# 37. The Equipment Selection Problem (RTG vs. RMG vs. Straddle)
## Tier 5: Integrated Digital Twin System-of-Systems Simulation

### Goal
Create a comprehensive digital twin that transforms equipment selection from static analysis to dynamic, real-time optimization through integrated system-of-systems modeling, enabling continuous what-if analysis and adaptive equipment deployment based on live operational feedback.

### Key Assumptions
- Digital twin can maintain bidirectional synchronization with physical terminal operations
- Real-time data from IoT sensors, equipment systems, and operational databases can be integrated
- Predictive analytics can forecast demand patterns and equipment performance degradation
- What-if scenarios can be simulated rapidly to support strategic equipment decisions

### Approach (Step-by-Step)
1. **System Architecture**: Design interconnected subsystems for equipment, traffic, weather, and financial modeling
2. **Real-Time Integration**: Implement data ingestion from IoT sensors, RFID, GPS, and operational systems
3. **Simulation Engine**: Create discrete event and continuous simulation models for terminal operations
4. **Predictive Analytics**: Develop machine learning models for demand forecasting and equipment performance
5. **Optimization Engine**: Implement real-time equipment allocation and deployment optimization
6. **What-If Analysis**: Enable rapid scenario testing for equipment configuration changes

### What to Look for in the Results
- Real-time monitoring dashboard with live KPI tracking
- Predictive analytics showing demand forecasts and equipment performance predictions
- What-if scenario analysis comparing different equipment configurations
- System-wide optimization showing measurable improvements over baseline operations

### Concrete Example
MCT's digital twin implementation with comprehensive system integration:
- **IoT Sensors**: 847 sensors across terminal processing 2.3M data points daily
- **Equipment Fleet**: 16 RTG cranes, 4 RMG cranes, 6 automated straddle carriers
- **Simulation Period**: 7-day operational trial with real-time synchronization
- **Optimization Impact**: 22% productivity increase, 34% downtime reduction, 12% cost savings
- **Scenarios**: Peak season, off-peak, vessel clustering, maintenance window optimization

### Why this Tier Exists vs. Previous Tiers
**Digital Twin provides:**
- **Real-Time Optimization**: Continuous adaptation based on live operational data
- **Predictive Capabilities**: Forecast demand and equipment maintenance needs
- **System Integration**: Holistic view of interconnected terminal operations
- **What-If Testing**: Rapid scenario analysis without physical implementation risks

**Advantages over Mathematical Optimization (Tier 1):**
- Handles dynamic, stochastic operational environments
- Incorporates real-time data and predictive analytics
- Models complex system interactions and emergent behaviors
- Provides continuous optimization rather than one-time decisions

**Advantages over Constraint Propagation (Tier 2):**
- Scales to complex, dynamic systems with thousands of variables

- Uses real operational data rather than synthetic simulations
- Provides real-time optimization rather than offline analysis
- Handles complex system dynamics and feedback loops
- Offers predictive and prescriptive analytics

**Advantages over Federated Learning (Tier 4):**
- Real-time optimization vs. historical pattern recognition
- Handles dynamic system interactions and feedback loops

**When to use this Tier:**
- Large terminal operations requiring real-time optimization

```python
# Import required libraries for digital twin simulation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from typing import List, Dict, Tuple, Optional, Callable
from dataclasses import dataclass, field
from datetime import datetime, timedelta
import random
import time
from collections import defaultdict, deque
import warnings
warnings.filterwarnings('ignore')

# Set plotting style for professional visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
@dataclass
class EquipmentUnit:
    """Individual equipment unit in the digital twin"""
    unit_id: str
    equipment_type: str  # RTG, RMG, SC
    location: Tuple[float, float]  # (x, y) coordinates
    status: str  # ACTIVE, MAINTENANCE, IDLE, FAULT
    productivity_mph: float  # Current productivity
    fuel_consumption: float  # Current fuel consumption
    maintenance_hours: float  # Hours until next maintenance
    utilization: float  # Current utilization (0-1)
    operating_cost_hour: float  # Operating cost per hour

@dataclass
class IoTSensor:
    """IoT sensor for real-time data collection"""
    sensor_id: str
    sensor_type: str  # EQUIPMENT, TRAFFIC, WEATHER, ENERGY
    location: Tuple[float, float]
    sampling_rate: int  # Seconds between readings
    last_reading: float = 0.0
    data_history: deque = field(default_factory=lambda: deque(maxlen=1000))

@dataclass
class VesselSchedule:
    """Vessel arrival and service schedule"""
    vessel_id: str
    eta: datetime  # Estimated time of arrival
    teu_capacity: int  # TEU capacity
    service_time_hours: float  # Expected service time
    priority: int  # Service priority (1=high, 5=low)
    berth_assigned: Optional[str] = None

@dataclass
class WeatherCondition:
    """Weather conditions affecting operations"""
    timestamp: datetime
    temperature: float  # Celsius
    wind_speed: float  # km/h
    visibility: float  # km
    precipitation: float  # mm/h
    impact_factor: float  # Impact on operations (0-1)

@dataclass
class DigitalTwinConfig:
    """Configuration for digital twin simulation"""
    simulation_hours: int = 168  # 1 week
    time_step_minutes: int = 15  # 15-minute intervals
    num_berths: int = 8
    num_blocks: int = 8
    yard_area_hectares: float = 50.0
    real_time_sync: bool = True
    prediction_horizon_hours: int = 24

print("Digital twin data structures defined")
print(f"Simulation will run for {DigitalTwinConfig().simulation_hours} hours with {DigitalTwinConfig().time_step_minutes} minute time steps")
```

```python
class DigitalTwinEquipmentSystem:
    """Equipment management system for digital twin"""
    
    def __init__(self, config: DigitalTwinConfig):
        self.config = config
        self.equipment_units = {}
        self.equipment_performance_history = defaultdict(list)
        self.maintenance_schedule = []
        
    def initialize_equipment_fleet(self, equipment_config: Dict[str, int]) -> None:
        """Initialize equipment fleet based on configuration"""
        unit_id_counter = 1
        
        for eq_type, num_units in equipment_config.items():
            for i in range(num_units):
                # Assign location based on equipment type
                if eq_type == 'RTG':
                    location = (random.uniform(0, config.yard_area_hectares), 
                               random.uniform(0, config.yard_area_hectares))
                    productivity = random.uniform(20, 24)  # moves/hour
                    fuel_consumption = random.uniform(15, 20)  # liters/hour
                    op_cost = random.uniform(45, 55)  # $/hour
                elif eq_type == 'RMG':
                    location = (random.uniform(0, config.yard_area_hectares), 
                               random.uniform(0, config.yard_area_hectares))
                    productivity = random.uniform(28, 32)  # moves/hour
                    fuel_consumption = random.uniform(5, 10)  # liters/hour (electric)
                    op_cost = random.uniform(35, 45)  # $/hour
                else:  # SC
                    location = (random.uniform(0, config.yard_area_hectares), 
                               random.uniform(0, config.yard_area_hectares))
                    productivity = random.uniform(16, 20)  # moves/hour
                    fuel_consumption = random.uniform(12, 18)  # liters/hour
                    op_cost = random.uniform(40, 50)  # $/hour
                
                unit = EquipmentUnit(
                    unit_id=f"{eq_type}_{unit_id_counter:03d}",
                    equipment_type=eq_type,
                    location=location,
                    status='ACTIVE',
                    productivity_mph=productivity,
                    fuel_consumption=fuel_consumption,
                    maintenance_hours=random.uniform(200, 500),
                    utilization=0.0,
                    operating_cost_hour=op_cost
                )
                
                self.equipment_units[unit.unit_id] = unit
                unit_id_counter += 1
        
        print(f"Initialized {len(self.equipment_units)} equipment units:")
        for eq_type in equipment_config:
            count = len([u for u in self.equipment_units.values() if u.equipment_type == eq_type])
            print(f"  {eq_type}: {count} units")
    
    def update_equipment_status(self, current_time: datetime, 
                               weather_impact: float, demand_factor: float) -> Dict[str, float]:
        """Update equipment status based on operating conditions"""
        kpi_data = {
            'total_moves': 0,
            'total_cost': 0,
            'active_units': 0,
            'maintenance_units': 0,
            'average_utilization': 0
        }
        
        for unit in self.equipment_units.values():
            # Update maintenance hours
            unit.maintenance_hours -= self.config.time_step_minutes / 60
            
            # Check if maintenance is required
            if unit.maintenance_hours <= 0:
                unit.status = 'MAINTENANCE'
                unit.maintenance_hours = random.uniform(50, 100)  # Maintenance duration
                self.maintenance_schedule.append({
                    'unit_id': unit.unit_id,
                    'start_time': current_time,
                    'duration_hours': unit.maintenance_hours
                })
            elif unit.status == 'MAINTENANCE' and unit.maintenance_hours <= 0:
                unit.status = 'ACTIVE'
                unit.maintenance_hours = random.uniform(200, 500)  # Next maintenance interval
            
            # Calculate performance based on conditions
            if unit.status == 'ACTIVE':
                # Weather and demand impact
                effective_productivity = unit.productivity_mph * (1 - weather_impact * 0.3)
                effective_productivity *= (0.5 + demand_factor * 0.5)  # Demand scaling
                
                # Random variations (equipment degradation, operator efficiency)
                performance_variation = random.uniform(0.9, 1.1)
                effective_productivity *= performance_variation
                
                # Update utilization
                unit.utilization = min(1.0, demand_factor + random.uniform(-0.1, 0.1))
                unit.utilization = max(0.0, unit.utilization)
                
                # Calculate moves and costs for this time step
                time_step_hours = self.config.time_step_minutes / 60
                moves = effective_productivity * unit.utilization * time_step_hours
                cost = unit.operating_cost_hour * unit.utilization * time_step_hours
                
                kpi_data['total_moves'] += moves
                kpi_data['total_cost'] += cost
                kpi_data['active_units'] += 1
                kpi_data['average_utilization'] += unit.utilization
                
                # Store performance history
                self.equipment_performance_history[unit.unit_id].append({
                    'timestamp': current_time,
                    'moves': moves,
                    'cost': cost,
                    'utilization': unit.utilization,
                    'productivity': effective_productivity
                })
                
            elif unit.status == 'MAINTENANCE':
                kpi_data['maintenance_units'] += 1
                # Maintenance costs
                time_step_hours = self.config.time_step_minutes / 60
                maintenance_cost = 50 * time_step_hours  # $50/hour maintenance cost
                kpi_data['total_cost'] += maintenance_cost
        
        # Calculate average utilization
        total_units = kpi_data['active_units'] + kpi_data['maintenance_units']
        if total_units > 0:
            kpi_data['average_utilization'] /= total_units
        
        return kpi_data
    
    def predict_equipment_performance(self, horizon_hours: int) -> Dict[str, List[float]]:
        """Predict equipment performance over future horizon"""
        predictions = {
            'moves_capacity': [],
            'operating_cost': [],
            'maintenance_events': []
            'utilization_forecast': []
        }
        
        for unit in self.equipment_units.values():
            # Simple linear prediction based on current performance
            recent_performance = self.equipment_performance_history[unit.unit_id][-10:] if len(self.equipment_performance_history[unit.unit_id]) >= 10 else self.equipment_performance_history[unit.unit_id]
            
            if recent_performance:
                avg_moves = np.mean([p['moves'] for p in recent_performance])
                avg_cost = np.mean([p['cost'] for p in recent_performance])
                avg_utilization = np.mean([p['utilization'] for p in recent_performance])
                
                # Predict maintenance events
                maintenance_risk = max(0, 1 - unit.maintenance_hours / 200)  # Risk increases as maintenance approaches
                
                # Add some randomness to predictions
                for hour in range(horizon_hours):
                    # Trend and seasonality
                    trend_factor = 1.0 + hour * 0.001  # Slight degradation trend
                    seasonal_factor = 0.9 + 0.2 * np.sin(hour * np.pi / 12)  # Daily seasonality
                    
                    predicted_moves = avg_moves * trend_factor * seasonal_factor * random.uniform(0.9, 1.1)
                    predicted_cost = avg_cost * trend_factor * seasonal_factor * random.uniform(0.9, 1.1)
                    predicted_utilization = avg_utilization * seasonal_factor * random.uniform(0.9, 1.1)
                    
                    if len(predictions['moves_capacity']) <= hour:
                        predictions['moves_capacity'][hour] += predicted_moves
                        predictions['operating_cost'][hour] += predicted_cost
                        predictions['utilization_forecast'][hour] += predicted_utilization
                        predictions['maintenance_events'][hour] += maintenance_risk
        
        return predictions

print("DigitalTwinEquipmentSystem class defined successfully")
```

```python
class DigitalTwinTrafficSystem:
    """Traffic and vessel management system for digital twin"""
    
    def __init__(self, config: DigitalTwinConfig):
        self.config = config
        self.vessel_schedule = []
        self.traffic_patterns = defaultdict(list)
        self.berth_utilization = defaultdict(float)
        
    def generate_vessel_schedule(self, num_vessels: int, start_time: datetime) -> None:
        """Generate realistic vessel arrival schedule"""
        self.vessel_schedule = []
        
        for i in range(num_vessels):
            # Random arrival within simulation period
            arrival_offset = random.uniform(0, self.config.simulation_hours)
            eta = start_time + timedelta(hours=arrival_offset)
            
            vessel = VesselSchedule(
                vessel_id=f"VESSEL_{i+1:03d}",
                eta=eta,
                teu_capacity=random.randint(1000, 8000),
                service_time_hours=random.uniform(12, 48),
                priority=random.randint(1, 5),
                berth_assigned=None
            )
            
            self.vessel_schedule.append(vessel)
        
        # Sort by ETA
        self.vessel_schedule.sort(key=lambda v: v.eta)
        
        print(f"Generated vessel schedule with {len(self.vessel_schedule)} vessels")
        print(f"First vessel ETA: {self.vessel_schedule[0].eta}")
        print(f"Last vessel ETA: {self.vessel_schedule[-1].eta}")
    
    def update_traffic_patterns(self, current_time: datetime) -> Dict[str, float]:
        """Update traffic patterns and calculate demand factors"""
        # Calculate demand based on vessel schedule
        active_vessels = 0
        total_teus = 0
        
        for vessel in self.vessel_schedule:
            # Check if vessel is currently being served
            service_start = vessel.eta
            service_end = vessel.eta + timedelta(hours=vessel.service_time_hours)
            
            if service_start <= current_time <= service_end:
                active_vessels += 1
                total_teus += vessel.teu_capacity
            
            # Check if vessel is arriving soon (next 2 hours)
            elif vessel.eta <= current_time + timedelta(hours=2):
                active_vessels += 0.5  # Partial weight for arriving vessels
                total_teus += vessel.teu_capacity * 0.5
        
        # Calculate demand factor (0-1 scale)
        max_teus = self.config.num_berths * 5000  # Maximum TEU capacity
        demand_factor = min(1.0, total_teus / max_teus)
        
        # Add time-of-day patterns
        hour_of_day = current_time.hour
        if 6 <= hour_of_day <= 18:  # Daytime
            demand_factor *= 1.2
        else:  # Nighttime
            demand_factor *= 0.8
        
        # Add day-of-week patterns
        day_of_week = current_time.weekday()
        if day_of_week >= 5:  # Weekend
            demand_factor *= 0.9
        else:  # Weekday
            demand_factor *= 1.1
        
        # Store traffic pattern
        self.traffic_patterns[current_time].append({
            'active_vessels': active_vessels,
            'total_teus': total_teus,
            'demand_factor': demand_factor
        })
        
        return {
            'active_vessels': active_vessels,
            'total_teus': total_teus,
            'demand_factor': min(1.0, demand_factor),
            'berth_utilization': active_vessels / self.config.num_berths
        }
    
    def predict_traffic_demand(self, horizon_hours: int, current_time: datetime) -> List[float]:
        """Predict traffic demand over future horizon"""
        demand_forecast = []
        
        for hour in range(horizon_hours):
            future_time = current_time + timedelta(hours=hour)
            
            # Count vessels arriving and being served
            arriving_vessels = 0
            serving_vessels = 0
            
            for vessel in self.vessel_schedule:
                service_start = vessel.eta
                service_end = vessel.eta + timedelta(hours=vessel.service_time_hours)
                
                if service_start <= future_time <= service_end:
                    serving_vessels += 1
                elif service_start <= future_time + timedelta(hours=2) >= service_start:
                    arriving_vessels += 1
            
            # Calculate predicted demand
            predicted_teus = (arriving_vessels + serving_vessels) * 3000  # Average TEU per vessel
            max_teus = self.config.num_berths * 5000
            predicted_demand = min(1.0, predicted_teus / max_teus)
            
            # Add time patterns
            hour_of_day = future_time.hour
            if 6 <= hour_of_day <= 18:
                predicted_demand *= 1.2
            else:
                predicted_demand *= 0.8
            
            demand_forecast.append(min(1.0, predicted_demand))
        
        return demand_forecast

print("DigitalTwinTrafficSystem class defined successfully")
```

```python
class DigitalTwinWeatherSystem:
    """Weather modeling system for digital twin"""
    
    def __init__(self, config: DigitalTwinConfig):
        self.config = config
        self.weather_history = []
        
    def generate_weather_sequence(self, start_time: datetime) -> None:
        """Generate realistic weather sequence for simulation period"""
        self.weather_history = []
        
        for hour in range(self.config.simulation_hours):
            current_time = start_time + timedelta(hours=hour)
            
            # Generate weather with realistic patterns
            base_temp = 20 + 10 * np.sin(hour * np.pi / 12)  # Daily temperature cycle
            temperature = base_temp + random.uniform(-5, 5)
            
            # Wind patterns
            wind_speed = random.uniform(5, 25) + random.uniform(-5, 5)
            wind_speed = max(0, wind_speed)
            
            # Visibility and precipitation
            visibility = random.uniform(5, 20)
            precipitation = max(0, random.uniform(-2, 5))  # Occasional rain
            
            # Calculate impact factor on operations
            impact_factor = 0.0
            
            # High wind impact
            if wind_speed > 30:
                impact_factor += 0.3
            elif wind_speed > 20:
                impact_factor += 0.15
            
            # Low visibility impact
            if visibility < 5:
                impact_factor += 0.2
            elif visibility < 10:
                impact_factor += 0.1
            
            # Precipitation impact
            if precipitation > 10:
                impact_factor += 0.25
            elif precipitation > 5:
                impact_factor += 0.1
            
            # Temperature extremes
            if temperature > 35 or temperature < 5:
                impact_factor += 0.1
            
            weather = WeatherCondition(
                timestamp=current_time,
                temperature=temperature,
                wind_speed=wind_speed,
                visibility=visibility,
                precipitation=precipitation,
                impact_factor=min(1.0, impact_factor)
            )
            
            self.weather_history.append(weather)
        
        print(f"Generated weather sequence for {len(self.weather_history)} hours")
        avg_impact = np.mean([w.impact_factor for w in self.weather_history])
        print(f"Average weather impact factor: {avg_impact:.3f}")
    
    def get_current_weather(self, current_time: datetime) -> WeatherCondition:
        """Get weather conditions for current time"""
        # Find closest weather record
        for weather in self.weather_history:
            if weather.timestamp <= current_time < weather.timestamp + timedelta(hours=1):
                return weather
        
        # Return default weather if not found
        return WeatherCondition(
            timestamp=current_time,
            temperature=20.0,
            wind_speed=10.0,
            visibility=15.0,
            precipitation=0.0,
            impact_factor=0.0
        )
    
    def predict_weather_impact(self, horizon_hours: int, current_time: datetime) -> List[float]:
        """Predict weather impact over future horizon"""
        impact_forecast = []
        
        for hour in range(horizon_hours):
            future_time = current_time + timedelta(hours=hour)
            weather = self.get_current_weather(future_time)
            impact_forecast.append(weather.impact_factor)
        
        return impact_forecast

print("DigitalTwinWeatherSystem class defined successfully")
```

```python
class IntegratedDigitalTwin:
    """Main digital twin system integrating all subsystems"""
    
    def __init__(self, config: DigitalTwinConfig):
        self.config = config
        self.equipment_system = DigitalTwinEquipmentSystem(config)
        self.traffic_system = DigitalTwinTrafficSystem(config)
        self.weather_system = DigitalTwinWeatherSystem(config)
        
        # Simulation state
       .current_time = None
       .simulation_history = []
       .kpi_history = defaultdict(list)
        
        # Optimization engine
        self.optimization_recommendations = []
        
    def initialize_simulation(self, equipment_config: Dict[str, int], 
                             start_time: datetime) -> None:
        """Initialize all subsystems for simulation"""
        self.current_time = start_time
        
        # Initialize equipment fleet
        self.equipment_system.initialize_equipment_fleet(equipment_config)
        
        # Generate vessel schedule
        num_vessels = random.randint(15, 25)  # 15-25 vessels per week
        self.traffic_system.generate_vessel_schedule(num_vessels, start_time)
        
        # Generate weather sequence
        self.weather_system.generate_weather_sequence(start_time)
        
        print(f"Digital twin initialized for simulation starting at {start_time}")
        print(f"Equipment config: {equipment_config}")
        print(f"Simulation duration: {self.config.simulation_hours} hours")
    
    def run_simulation_step(self) -> Dict[str, float]:
        """Run one simulation time step"""
        # Get current conditions
        weather = self.weather_system.get_current_weather(self.current_time)
        traffic_data = self.traffic_system.update_traffic_patterns(self.current_time)
        equipment_kpi = self.equipment_system.update_equipment_status(
            self.current_time, weather.impact_factor, traffic_data['demand_factor']
        )
        
        # Calculate comprehensive KPIs
        step_kpis = {
            'timestamp': self.current_time,
            'total_moves': equipment_kpi['total_moves'],
            'total_cost': equipment_kpi['total_cost'],
            'cost_per_move': equipment_kpi['total_cost'] / max(1, equipment_kpi['total_moves']),
            'equipment_utilization': equipment_kpi['average_utilization'],
            'active_equipment': equipment_kpi['active_units'],
            'maintenance_equipment': equipment_kpi['maintenance_units'],
            'weather_impact': weather.impact_factor,
            'demand_factor': traffic_data['demand_factor'],
            'berth_utilization': traffic_data['berth_utilization'],
            'vessel_count': traffic_data['active_vessels']
        }
        
        # Store in history
        self.simulation_history.append(step_kpis)
        for key, value in step_kpis.items():
            if key != 'timestamp' and isinstance(value, (int, float)):
                self.kpi_history[key].append(value)
        
        # Advance time
        self.current_time += timedelta(minutes=self.config.time_step_minutes)
        
        return step_kpis
    
    def run_complete_simulation(self) -> Dict[str, List[float]]:
        """Run complete simulation for configured duration"""
        print(f"\n=== STARTING DIGITAL TWIN SIMULATION ===")
        print(f"Simulation period: {self.config.simulation_hours} hours")
        print(f"Time step: {self.config.time_step_minutes} minutes")
        
        start_time = time.time()
        
        # Calculate number of time steps
        total_steps = (self.config.simulation_hours * 60) // self.config.time_step_minutes
        
        for step in range(total_steps):
            step_kpis = self.run_simulation_step()
            
            # Progress reporting
            if (step + 1) % (total_steps // 10) == 0:
                progress = (step + 1) / total_steps * 100
                print(f"Progress: {progress:.1f}% - Time: {step_kpis['timestamp']}")
        
        end_time = time.time()
        print(f"\nSimulation completed in {end_time - start_time:.2f} seconds")
        print(f"Total simulation steps: {len(self.simulation_history)}")
        
        return dict(self.kpi_history)
    
    def generate_optimization_recommendations(self) -> List[Dict]:
        """Generate optimization recommendations based on simulation results"""
        recommendations = []
        
        # Analyze equipment utilization
        avg_utilization = np.mean(self.kpi_history['equipment_utilization'])
        low_utilization_threshold = 0.6
        
        if avg_utilization < low_utilization_threshold:
            recommendations.append({
                'type': 'EQUIPMENT_OPTIMIZATION',
                'priority': 'HIGH',
                'description': f'Low equipment utilization ({avg_utilization:.1%}) - consider fleet optimization',
                'potential_savings': (1 - avg_utilization) * 100,
                'action': 'Optimize equipment allocation or reduce fleet size'
            })
        
        # Analyze cost efficiency
        avg_cost_per_move = np.mean(self.kpi_history['cost_per_move'])
        if avg_cost_per_move > 50:  # High cost threshold
            recommendations.append({
                'type': 'COST_OPTIMIZATION',
                'priority': 'MEDIUM',
                'description': f'High cost per move (${avg_cost_per_move:.2f}) - investigate efficiency improvements',
                'potential_savings': (avg_cost_per_move - 50) / avg_cost_per_move * 100,
                'action': 'Review equipment mix and operational procedures'
            })
        
        # Analyze weather impact
        avg_weather_impact = np.mean(self.kpi_history['weather_impact'])
        if avg_weather_impact > 0.2:
            recommendations.append({
                'type': 'WEATHER_RESILIENCE',
                'priority': 'MEDIUM',
                'description': f'Significant weather impact ({avg_weather_impact:.1%}) - improve weather resilience',
                'potential_improvement': avg_weather_impact * 50,
                'action': 'Invest in weather-resistant equipment and procedures'
            })
        
        # Analyze maintenance patterns
        maintenance_events = len(self.equipment_system.maintenance_schedule)
        if maintenance_events > len(self.equipment_system.equipment_units) * 0.3:
            recommendations.append({
                'type': 'MAINTENANCE_OPTIMIZATION',
                'priority': 'HIGH',
                'description': f'High maintenance frequency ({maintenance_events} events) - optimize maintenance schedule',
                'potential_improvement': 20,
                'action': 'Implement predictive maintenance and optimize scheduling'
            })
        
        self.optimization_recommendations = recommendations
        return recommendations
    
    def run_what_if_scenario(self, scenario_name: str, 
                           equipment_modification: Dict[str, int]) -> Dict[str, float]:
        """Run what-if scenario with modified equipment configuration"""
        print(f"\n=== WHAT-IF SCENARIO: {scenario_name} ===")
        
        # Save current state
        original_time = self.current_time
        original_history = self.simulation_history.copy()
        original_kpi = dict(self.kpi_history)
        
        # Reset and reinitialize with modified equipment
        self.current_time = original_time - timedelta(hours=self.config.simulation_hours)
        self.simulation_history = []
        self.kpi_history = defaultdict(list)
        
        # Reinitialize with modified equipment
        start_time = self.current_time
        self.initialize_simulation(equipment_modification, start_time)
        
        # Run scenario simulation
        scenario_kpis = self.run_complete_simulation()
        
        # Calculate scenario summary
        scenario_summary = {
            'total_moves': np.sum(scenario_kpis['total_moves']),
            'total_cost': np.sum(scenario_kpis['total_cost']),
            'avg_cost_per_move': np.mean(scenario_kpis['cost_per_move']),
            'avg_utilization': np.mean(scenario_kpis['equipment_utilization']),
            'avg_weather_impact': np.mean(scenario_kpis['weather_impact']),
            'peak_demand': np.max(scenario_kpis['demand_factor'])
        }
        
        # Restore original state
        self.current_time = original_time
        self.simulation_history = original_history
        self.kpi_history = original_kpi
        
        return scenario_summary

print("IntegratedDigitalTwin class defined successfully")
```
