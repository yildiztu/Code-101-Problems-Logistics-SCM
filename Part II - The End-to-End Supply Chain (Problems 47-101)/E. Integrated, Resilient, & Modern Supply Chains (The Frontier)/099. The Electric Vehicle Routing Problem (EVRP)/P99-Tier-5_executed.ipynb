# Problem 99: The Electric Vehicle Routing Problem (EVRP)

## Tier 5: Integrated Digital Twin (Fleet-Wide Energy Simulation)

### Key Assumptions
- Real-time synchronization between physical and virtual EV fleet systems
- IoT sensors provide continuous data streams from vehicles and infrastructure
- Digital twin can simulate and predict energy consumption with high accuracy
- System-of-systems integration enables comprehensive fleet optimization
- Predictive analytics support proactive decision making and what-if analysis

### Approach (Step-by-Step)

The Integrated Digital Twin creates a living virtual replica of the entire EV logistics ecosystem:

1. **Physical Asset Modeling**: Create digital counterparts for vehicles, charging stations, and infrastructure
2. **IoT Data Integration**: Simulate real-time sensor data from vehicles and charging stations
3. **Energy Simulation**: Model battery dynamics, charging behavior, and consumption patterns
4. **Real-time Synchronization**: Maintain 1-second update cycles between physical and virtual systems
5. **Predictive Analytics**: Forecast energy needs, charging demand, and route optimization
6. **What-if Analysis**: Test scenarios like new charging stations or vehicle additions
7. **System Optimization**: Continuously re-optimize based on real-time conditions
8. **Performance Monitoring**: Track KPIs and system health across the entire fleet

### What to Look for in the Results
- Real-time monitoring dashboard with live fleet status
- Energy consumption patterns and predictive analytics
- Scenario analysis showing impact of strategic decisions
- System synchronization accuracy and latency metrics
- Comprehensive KPI tracking and performance optimization

### Concrete Example

We'll build a comprehensive digital twin for a mid-sized EV fleet:
- 1 depot
- 6 customers with dynamic demand patterns
- 3 charging stations with real-time availability
- 4 electric vehicles with IoT sensor integration
- Battery capacity: 100 kWh
- Energy consumption: 0.8 kWh per km (variable based on conditions)
- Real-time simulation with 500+ IoT sensors

```python
try:
    @dataclass
    class IoTDataPoint:
        """Represents a single IoT data point"""
        timestamp: datetime
        sensor_id: str
        value: float
        unit: str
        quality_score: float  # Data quality (0-1)
        
    @dataclass
    class VehicleTwin:
        """Digital twin of an electric vehicle"""
        vehicle_id: int
        location: Tuple[float, float]
        battery_level: float  # kWh
        battery_health: float  # 0-1 (degradation factor)
        speed: float  # km/h
        payload: float  # kg
        temperature: float  # Celsius
        energy_consumption_rate: float  # kWh/km
        route: List[int] = field(default_factory=list)
        sensors: Dict[str, float] = field(default_factory=dict)
        last_update: datetime = field(default_factory=datetime.now)
        
    @dataclass
    class ChargingStationTwin:
        """Digital twin of a charging station"""
        station_id: int
        location: Tuple[float, float]
        capacity: int  # Number of charging points
        available_points: int
        charging_rate: float  # kW
        queue_length: int
        utilization_rate: float  # 0-1
        energy_price: float  # $/kWh
        grid_carbon_intensity: float  # gCO2/kWh
        sensors: Dict[str, float] = field(default_factory=dict)
        last_update: datetime = field(default_factory=datetime.now)
        
    @dataclass
    class EnvironmentTwin:
        """Digital twin of the environment"""
        temperature: float  # Celsius
        humidity: float  # %
        wind_speed: float  # km/h
        traffic_density: Dict[str, float]  # Area -> density
        road_conditions: Dict[str, str]  # Area -> condition
        carbon_intensity: float  # gCO2/kWh (grid average)
        
    @dataclass
    class DigitalTwinSystem:
        """Complete digital twin system for EV fleet"""
        vehicles: List[VehicleTwin]
        charging_stations: List[ChargingStationTwin]
        environment: EnvironmentTwin
        iot_data_stream: deque
        synchronization_interval: float  # seconds
        prediction_horizon: timedelta
        kpi_history: Dict[str, List[float]]
        
    def create_digital_twin_instance():
        """Create a comprehensive digital twin instance"""
        
        # Create vehicle twins
        vehicles = [
            VehicleTwin(
                vehicle_id=0,
                location=(0, 0),  # Start at depot
                battery_level=95.0,
                battery_health=0.98,
                speed=0.0,
                payload=500.0,
                temperature=22.0,
                energy_consumption_rate=0.8,
                sensors={
                    'gps_accuracy': 0.95,
                    'battery_sensor_accuracy': 0.99,
                    'temperature_sensor': 22.5,
                    'motor_efficiency': 0.92,
                    'tire_pressure': 32.5
                }
            ),
            VehicleTwin(
                vehicle_id=1,
                location=(0, 0),
                battery_level=88.0,
                battery_health=0.96,
                speed=0.0,
                payload=450.0,
                temperature=21.8,
                energy_consumption_rate=0.78,
                sensors={
                    'gps_accuracy': 0.94,
                    'battery_sensor_accuracy': 0.98,
                    'temperature_sensor': 21.5,
                    'motor_efficiency': 0.90,
                    'tire_pressure': 31.8
                }
            ),
            VehicleTwin(
                vehicle_id=2,
                location=(0, 0),
                battery_level=92.0,
                battery_health=0.99,
                speed=0.0,
                payload=520.0,
                temperature=22.3,
                energy_consumption_rate=0.82,
                sensors={
                    'gps_accuracy': 0.96,
                    'battery_sensor_accuracy': 0.99,
                    'temperature_sensor': 22.8,
                    'motor_efficiency': 0.93,
                    'tire_pressure': 33.1
                }
            ),
            VehicleTwin(
                vehicle_id=3,
                location=(0, 0),
                battery_level=85.0,
                battery_health=0.97,
                speed=0.0,
                payload=480.0,
                temperature=21.9,
                energy_consumption_rate=0.79,
                sensors={
                    'gps_accuracy': 0.95,
                    'battery_sensor_accuracy': 0.98,
                    'temperature_sensor': 21.7,
                    'motor_efficiency': 0.91,
                    'tire_pressure': 32.2
                }
            )
        ]
        
        # Create charging station twins
        charging_stations = [
            ChargingStationTwin(
                station_id=0,
                location=(8, 12),
                capacity=4,
                available_points=4,
                charging_rate=50.0,
                queue_length=0,
                utilization_rate=0.0,
                energy_price=0.25,
                grid_carbon_intensity=150.0,
                sensors={
                    'power_output': 49.8,
                    'efficiency': 0.94,
                    'availability': 1.0,
                    'connector_status': [1, 1, 1, 1],
                    'temperature': 25.3
                }
            ),
            ChargingStationTwin(
                station_id=1,
                location=(16, 8),
                capacity=3,
                available_points=3,
                charging_rate=75.0,
                queue_length=0,
                utilization_rate=0.0,
                energy_price=0.28,
                grid_carbon_intensity=165.0,
                sensors={
                    'power_output': 74.5,
                    'efficiency': 0.96,
                    'availability': 1.0,
                    'connector_status': [1, 1, 1],
                    'temperature': 26.1
                }
            ),
            ChargingStationTwin(
                station_id=2,
                location=(12, 18),
                capacity=2,
                available_points=2,
                charging_rate=100.0,
                queue_length=0,
                utilization_rate=0.0,
                energy_price=0.32,
                grid_carbon_intensity=180.0,
                sensors={
                    'power_output': 99.2,
                    'efficiency': 0.97,
                    'availability': 1.0,
                    'connector_status': [1, 1],
                    'temperature': 27.2
                }
            )
        ]
        
        # Create environment twin
        environment = EnvironmentTwin(
            temperature=22.0,
            humidity=65.0,
            wind_speed=12.0,
            traffic_density={
                'downtown': 0.8,
                'suburban': 0.4,
                'industrial': 0.6,
                'highway': 0.3
            },
            road_conditions={
                'downtown': 'moderate',
                'suburban': 'good',
                'industrial': 'poor',
                'highway': 'good'
            },
            carbon_intensity=160.0
        )
        
        # Create digital twin system
        digital_twin = DigitalTwinSystem(
            vehicles=vehicles,
            charging_stations=charging_stations,
            environment=environment,
            iot_data_stream=deque(maxlen=10000),
            synchronization_interval=1.0,  # 1-second updates
            prediction_horizon=timedelta(hours=2),
            kpi_history={
                'total_energy_consumed': [],
                'average_battery_level': [],
                'fleet_utilization': [],
                'charging_station_utilization': [],
                'carbon_emissions': [],
                'operational_costs': []
                'service_level': []
                'prediction_accuracy': []
            }
        )
        
        return digital_twin
    
    # Create digital twin instance
    digital_twin = create_digital_twin_instance()
    print(f"Created digital twin with:")
    print(f"  {len(digital_twin.vehicles)} vehicle twins")
    print(f"  {len(digital_twin.charging_stations)} charging station twins")
    print(f"  Synchronization interval: {digital_twin.synchronization_interval} seconds")
    print(f"  Prediction horizon: {digital_twin.prediction_horizon}")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    class IoTDataSimulator:
        """Simulates IoT sensor data from vehicles and infrastructure"""
        
        def __init__(self, digital_twin: DigitalTwinSystem):
            self.digital_twin = digital_twin
            self.sensor_error_rate = 0.02  # 2% sensor error rate
            self.data_latency = 0.1  # 100ms average latency
            
        def generate_vehicle_sensor_data(self, vehicle: VehicleTwin, 
                                       timestamp: datetime) -> List[IoTDataPoint]:
            """Generate sensor data for a vehicle"""
            
            data_points = []
            
            # GPS location data
            gps_error = np.random.normal(0, 0.5) if np.random.random() > self.sensor_error_rate else np.random.normal(0, 5)
            data_points.append(IoTDataPoint(
                timestamp=timestamp,
                sensor_id=f"vehicle_{vehicle.vehicle_id}_gps",
                value=gps_error,
                unit="m",
                quality_score=0.95 if np.random.random() > self.sensor_error_rate else 0.7
            ))
            
            # Battery level sensor
            battery_noise = np.random.normal(0, 0.5) if np.random.random() > self.sensor_error_rate else np.random.normal(0, 2)
            data_points.append(IoTDataPoint(
                timestamp=timestamp,
                sensor_id=f"vehicle_{vehicle.vehicle_id}_battery",
                value=vehicle.battery_level + battery_noise,
                unit="kWh",
                quality_score=0.98 if np.random.random() > self.sensor_error_rate else 0.8
            ))
            
            # Speed sensor
            speed_noise = np.random.normal(0, 1) if np.random.random() > self.sensor_error_rate else np.random.normal(0, 5)
            data_points.append(IoTDataPoint(
                timestamp=timestamp,
                sensor_id=f"vehicle_{vehicle.vehicle_id}_speed",
                value=max(0, vehicle.speed + speed_noise),
                unit="km/h",
                quality_score=0.96 if np.random.random() > self.sensor_error_rate else 0.75
            ))
            
            # Energy consumption rate
            consumption_variation = np.random.normal(0, 0.02) if np.random.random() > self.sensor_error_rate else np.random.normal(0, 0.1)
            data_points.append(IoTDataPoint(
                timestamp=timestamp,
                sensor_id=f"vehicle_{vehicle.vehicle_id}_consumption",
                value=max(0, vehicle.energy_consumption_rate + consumption_variation),
                unit="kWh/km",
                quality_score=0.94 if np.random.random() > self.sensor_error_rate else 0.7
            ))
            
            # Temperature sensor
            temp_variation = np.random.normal(0, 0.5) if np.random.random() > self.sensor_error_rate else np.random.normal(0, 2)
            data_points.append(IoTDataPoint(
                timestamp=timestamp,
                sensor_id=f"vehicle_{vehicle.vehicle_id}_temperature",
                value=vehicle.temperature + temp_variation,
                unit="°C",
                quality_score=0.97 if np.random.random() > self.sensor_error_rate else 0.8
            ))
            
            return data_points
        
        def generate_charging_station_data(self, station: ChargingStationTwin, 
                                         timestamp: datetime) -> List[IoTDataPoint]:
            """Generate sensor data for a charging station"""
            
            data_points = []
            
            # Power output sensor
            power_variation = np.random.normal(0, 1) if np.random.random() > self.sensor_error_rate else np.random.normal(0, 5)
            actual_power = station.charging_rate * (station.capacity - station.available_points)
            data_points.append(IoTDataPoint(
                timestamp=timestamp,
                sensor_id=f"station_{station.station_id}_power",
                value=max(0, actual_power + power_variation),
                unit="kW",
                quality_score=0.96 if np.random.random() > self.sensor_error_rate else 0.8
            ))
            
            # Availability sensor
            data_points.append(IoTDataPoint(
                timestamp=timestamp,
                sensor_id=f"station_{station.station_id}_availability",
                value=station.available_points,
                unit="points",
                quality_score=0.99 if np.random.random() > self.sensor_error_rate else 0.9
            ))
            
            # Queue length sensor
            data_points.append(IoTDataPoint(
                timestamp=timestamp,
                sensor_id=f"station_{station.station_id}_queue",
                value=station.queue_length,
                unit="vehicles",
                quality_score=0.98 if np.random.random() > self.sensor_error_rate else 0.85
            ))
            
            return data_points
        
        def generate_environmental_data(self, timestamp: datetime) -> List[IoTDataPoint]:
            """Generate environmental sensor data"""
            
            data_points = []
            
            # Temperature
            temp_variation = np.random.normal(0, 0.2)
            data_points.append(IoTDataPoint(
                timestamp=timestamp,
                sensor_id="env_temperature",
                value=self.digital_twin.environment.temperature + temp_variation,
                unit="°C",
                quality_score=0.97
            ))
            
            # Grid carbon intensity
            carbon_variation = np.random.normal(0, 5)
            data_points.append(IoTDataPoint(
                timestamp=timestamp,
                sensor_id="grid_carbon",
                value=max(0, self.digital_twin.environment.carbon_intensity + carbon_variation),
                unit="gCO2/kWh",
                quality_score=0.95
            ))
            
            return data_points
        
        def simulate_iot_stream(self, duration_seconds: int = 60):
            """Simulate IoT data stream for specified duration"""
            
            start_time = datetime.now()
            data_points_generated = 0
            
            for second in range(duration_seconds):
                timestamp = start_time + timedelta(seconds=second)
                
                # Generate vehicle data
                for vehicle in self.digital_twin.vehicles:
                    vehicle_data = self.generate_vehicle_sensor_data(vehicle, timestamp)
                    for data_point in vehicle_data:
                        self.digital_twin.iot_data_stream.append(data_point)
                        data_points_generated += 1
                
                # Generate charging station data
                for station in self.digital_twin.charging_stations:
                    station_data = self.generate_charging_station_data(station, timestamp)
                    for data_point in station_data:
                        self.digital_twin.iot_data_stream.append(data_point)
                        data_points_generated += 1
                
                # Generate environmental data
                env_data = self.generate_environmental_data(timestamp)
                for data_point in env_data:
                    self.digital_twin.iot_data_stream.append(data_point)
                    data_points_generated += 1
            
            return data_points_generated
    
    # Create IoT data simulator
    iot_simulator = IoTDataSimulator(digital_twin)
    print(f"Created IoT data simulator with {iot_simulator.sensor_error_rate:.1%} error rate")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    class EnergySimulationEngine:
        """Advanced energy simulation for EV fleet"""
        
        def __init__(self, digital_twin: DigitalTwinSystem):
            self.digital_twin = digital_twin
            
        def calculate_energy_consumption(self, vehicle: VehicleTwin, 
                                       distance: float, traffic_factor: float, 
                                       temperature_factor: float) -> float:
            """Calculate realistic energy consumption"""
            
            base_consumption = vehicle.energy_consumption_rate * distance
            
            # Traffic impact (higher consumption in heavy traffic)
            traffic_impact = 1.0 + (traffic_factor - 0.5) * 0.3
            
            # Temperature impact (battery efficiency varies with temperature)
            optimal_temp = 20.0  # Optimal battery temperature
            temp_diff = abs(vehicle.temperature - optimal_temp)
            temp_impact = 1.0 + (temp_diff / 20.0) * 0.15
            
            # Payload impact
            payload_impact = 1.0 + (vehicle.payload / 1000.0) * 0.1
            
            # Battery health impact
            health_impact = 1.0 + (1.0 - vehicle.battery_health) * 0.2
            
            # Speed impact (optimal speed around 50-60 km/h)
            optimal_speed = 55.0
            speed_diff = abs(vehicle.speed - optimal_speed)
            speed_impact = 1.0 + (speed_diff / 50.0) * 0.1
            
            total_consumption = (base_consumption * traffic_impact * temp_impact * 
                               payload_impact * health_impact * speed_impact)
            
            return total_consumption
        
        def simulate_charging_behavior(self, vehicle: VehicleTwin, 
                                      station: ChargingStationTwin, 
                                      charging_time: float) -> Dict[str, float]:
            """Simulate vehicle charging behavior"""
            
            # Calculate actual charging rate (varies with temperature and battery level)
            battery_level_factor = 1.0 - (vehicle.battery_level / 100.0) * 0.3  # Slower charging when nearly full
            temperature_factor = 1.0 - abs(vehicle.temperature - 25.0) / 50.0  # Optimal at 25°C
            
            actual_rate = station.charging_rate * battery_level_factor * temperature_factor * station.sensors.get('efficiency', 0.95)
            
            # Energy added
            energy_added = min(actual_rate * charging_time / 3600.0, 
                              vehicle.battery_capacity - vehicle.battery_level)
            
            # Cost calculation
            cost = energy_added * station.energy_price
            
            # Carbon emissions
            carbon_emissions = energy_added * station.grid_carbon_intensity / 1000.0  # kg CO2
            
            return {
                'energy_added': energy_added,
                'actual_rate': actual_rate,
                'cost': cost,
                'carbon_emissions': carbon_emissions,
                'final_battery_level': vehicle.battery_level + energy_added
            }
        
        def predict_energy_needs(self, vehicle: VehicleTwin, 
                               planned_route: List[Tuple[float, float]], 
                               time_horizon: timedelta) -> Dict[str, Any]:
            """Predict energy needs for planned route"""
            
            total_distance = 0
            total_energy_needed = 0
            charging_stops_needed = 0
            
            # Calculate route distance
            for i in range(len(planned_route) - 1):
                total_distance += np.sqrt(
                    (planned_route[i+1][0] - planned_route[i][0])**2 + 
                    (planned_route[i+1][1] - planned_route[i][1])**2
                )
            
            # Estimate energy consumption
            avg_traffic_factor = np.mean(list(self.digital_twin.environment.traffic_density.values()))
            temp_factor = 1.0 + abs(vehicle.temperature - 20.0) / 20.0 * 0.15
            
            estimated_consumption = self.calculate_energy_consumption(
                vehicle, total_distance, avg_traffic_factor, temp_factor
            )
            
            # Check if charging is needed
            if estimated_consumption > vehicle.battery_level * 0.8:  # 20% safety margin
                charging_stops_needed = max(1, int(estimated_consumption / (vehicle.battery_capacity * 0.6)))
            
            return {
                'total_distance': total_distance,
                'estimated_energy': estimated_consumption,
                'charging_stops_needed': charging_stops_needed,
                'can_complete_route': estimated_consumption <= vehicle.battery_level,
                'recommended_charging': max(0, estimated_consumption - vehicle.battery_level * 0.8)
            }
    
    # Create energy simulation engine
    energy_simulator = EnergySimulationEngine(digital_twin)
    print(f"Created energy simulation engine")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    class PredictiveAnalytics:
        """Predictive analytics for digital twin"""
        
        def __init__(self, digital_twin: DigitalTwinSystem):
            self.digital_twin = digital_twin
            self.prediction_models = {}
            
        def predict_charging_demand(self, time_horizon: timedelta) -> Dict[str, float]:
            """Predict charging station demand"""
            
            predictions = {}
            
            for station in self.digital_twin.charging_stations:
                # Simple prediction based on current utilization and time patterns
                base_demand = station.utilization_rate
                
                # Time-based variation (higher demand during peak hours)
                current_hour = datetime.now().hour
                if 8 <= current_hour <= 18:  # Business hours
                    time_multiplier = 1.5
                elif 18 <= current_hour <= 22:  # Evening
                    time_multiplier = 1.2
                else:  # Night
                    time_multiplier = 0.6
                
                predicted_demand = min(1.0, base_demand * time_multiplier)
                predictions[f"station_{station.station_id}"] = predicted_demand
            
            return predictions
        
        def predict_battery_degradation(self, vehicle: VehicleTwin, 
                                     days_ahead: int = 30) -> Dict[str, float]:
            """Predict battery degradation"""
            
            current_health = vehicle.battery_health
            daily_degradation_rate = 0.0001  # 0.01% per day under normal usage
            
            # Adjust for usage patterns
            usage_factor = 1.0 + (vehicle.speed / 50.0) * 0.2  # Higher speed = more stress
            temp_factor = 1.0 + abs(vehicle.temperature - 20.0) / 30.0 * 0.3  # Temperature stress
            
            adjusted_degradation_rate = daily_degradation_rate * usage_factor * temp_factor
            
            predicted_health = current_health - (adjusted_degradation_rate * days_ahead)
            
            return {
                'current_health': current_health,
                'predicted_health': max(0.7, predicted_health),  # Minimum 70% health
                'degradation_rate': adjusted_degradation_rate,
                'days_to_80_percent': max(0, int((current_health - 0.8) / adjusted_degradation_rate))
            }
        
        def predict_optimal_charging_times(self, vehicle: VehicleTwin) -> Dict[str, Any]:
            """Predict optimal charging times based on prices and carbon intensity"""
            
            current_hour = datetime.now().hour
            optimal_times = []
            
            # Analyze next 24 hours
            for hour in range(24):
                check_hour = (current_hour + hour) % 24
                
                # Price patterns (higher during peak hours)
                if 8 <= check_hour <= 18:
                    price_multiplier = 1.3
                elif 18 <= check_hour <= 22:
                    price_multiplier = 1.1
                else:
                    price_multiplier = 0.8
                
                # Carbon intensity patterns (lower during renewable hours)
                if 10 <= check_hour <= 16:  # Solar peak
                    carbon_multiplier = 0.7
                elif 0 <= check_hour <= 6:  # Wind peak
                    carbon_multiplier = 0.8
                else:
                    carbon_multiplier = 1.2
                
                # Calculate score (lower is better)
                avg_price = np.mean([s.energy_price for s in self.digital_twin.charging_stations])
                avg_carbon = np.mean([s.grid_carbon_intensity for s in self.digital_twin.charging_stations])
                
                price_score = avg_price * price_multiplier
                carbon_score = avg_carbon * carbon_multiplier
                
                total_score = price_score * 0.6 + carbon_score * 0.4  # Weighted combination
                
                optimal_times.append({
                    'hour': check_hour,
                    'price_multiplier': price_multiplier,
                    'carbon_multiplier': carbon_multiplier,
                    'total_score': total_score,
                    'rank': 0  # Will be filled later
                })
            
            # Rank by total score
            optimal_times.sort(key=lambda x: x['total_score'])
            for i, time_info in enumerate(optimal_times):
                time_info['rank'] = i + 1
            
            return {
                'optimal_times': optimal_times[:6],  # Top 6 optimal times
                'best_hour': optimal_times[0]['hour'],
                'worst_hour': optimal_times[-1]['hour']
            }
    
    # Create predictive analytics
    predictive_analytics = PredictiveAnalytics(digital_twin)
    print(f"Created predictive analytics module")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    class DigitalTwinOrchestrator:
        """Orchestrates the entire digital twin system"""
        
        def __init__(self, digital_twin: DigitalTwinSystem):
            self.digital_twin = digital_twin
            self.iot_simulator = IoTDataSimulator(digital_twin)
            self.energy_simulator = EnergySimulationEngine(digital_twin)
            self.predictive_analytics = PredictiveAnalytics(digital_twin)
            
            # Simulation state
            self.current_time = datetime.now()
            self.simulation_step = 0
            
        def synchronize_physical_to_virtual(self, timestamp: datetime):
            """Synchronize physical assets with virtual twins"""
            
            # Process recent IoT data
            recent_data = [dp for dp in self.digital_twin.iot_data_stream 
                          if timestamp - dp.timestamp <= timedelta(seconds=2)]
            
            # Update vehicle twins
            for vehicle in self.digital_twin.vehicles:
                vehicle_data = [dp for dp in recent_data if f"vehicle_{vehicle.vehicle_id}_" in dp.sensor_id]
                
                for data_point in vehicle_data:
                    if "gps" in data_point.sensor_id:
                        # Update location (simplified)
                        pass  # Would update actual coordinates
                    elif "battery" in data_point.sensor_id:
                        # Update battery level
                        vehicle.battery_level = max(0, min(100, data_point.value))
                    elif "speed" in data_point.sensor_id:
                        # Update speed
                        vehicle.speed = max(0, data_point.value)
                    elif "consumption" in data_point.sensor_id:
                        # Update consumption rate
                        vehicle.energy_consumption_rate = max(0, data_point.value)
                    elif "temperature" in data_point.sensor_id:
                        # Update temperature
                        vehicle.temperature = data_point.value
                
                vehicle.last_update = timestamp
            
            # Update charging station twins
            for station in self.digital_twin.charging_stations:
                station_data = [dp for dp in recent_data if f"station_{station.station_id}_" in dp.sensor_id]
                
                for data_point in station_data:
                    if "availability" in data_point.sensor_id:
                        station.available_points = int(data_point.value)
                    elif "queue" in data_point.sensor_id:
                        station.queue_length = int(data_point.value)
                    elif "power" in data_point.sensor_id:
                        # Update utilization based on power output
                        max_power = station.charging_rate * station.capacity
                        station.utilization_rate = min(1.0, data_point.value / max_power)
                
                station.last_update = timestamp
        
        def calculate_kpis(self, timestamp: datetime) -> Dict[str, float]:
            """Calculate current KPIs"""
            
            # Energy KPIs
            total_energy = sum(v.battery_capacity * (1 - v.battery_level/100) for v in self.digital_twin.vehicles)
            avg_battery = np.mean([v.battery_level for v in self.digital_twin.vehicles])
            
            # Fleet utilization
            active_vehicles = sum(1 for v in self.digital_twin.vehicles if v.speed > 0)
            fleet_utilization = active_vehicles / len(self.digital_twin.vehicles)
            
            # Charging station utilization
            avg_charging_utilization = np.mean([s.utilization_rate for s in self.digital_twin.charging_stations])
            
            # Carbon emissions
            total_carbon = 0
            for vehicle in self.digital_twin.vehicles:
                energy_used = vehicle.battery_capacity * (1 - vehicle.battery_level/100)
                total_carbon += energy_used * self.digital_twin.environment.carbon_intensity / 1000
            
            # Operational costs
            total_cost = 0
            for station in self.digital_twin.charging_stations:
                energy_used = station.charging_rate * (station.capacity - station.available_points) * 0.1  # Simplified
                total_cost += energy_used * station.energy_price
            
            # Service level (simplified)
            service_level = 0.95  # Would be calculated based on actual deliveries
            
            return {
                'total_energy_consumed': total_energy,
                'average_battery_level': avg_battery,
                'fleet_utilization': fleet_utilization,
                'charging_station_utilization': avg_charging_utilization,
                'carbon_emissions': total_carbon,
                'operational_costs': total_cost,
                'service_level': service_level
            }
        
        def run_simulation_step(self, duration_minutes: int = 60):
            """Run one simulation step"""
            
            step_start = time.time()
            
            # Generate IoT data for the duration
            data_points = self.iot_simulator.simulate_iot_stream(duration_minutes)
            
            # Process synchronization at each second
            for second in range(duration_minutes):
                timestamp = self.current_time + timedelta(seconds=second)
                self.synchronize_physical_to_virtual(timestamp)
                
                # Calculate and store KPIs
                if second % 10 == 0:  # Every 10 seconds
                    kpis = self.calculate_kpis(timestamp)
                    for key, value in kpis.items():
                        self.digital_twin.kpi_history[key].append(value)
            
            # Update simulation time
            self.current_time += timedelta(minutes=duration_minutes)
            self.simulation_step += 1
            
            step_duration = time.time() - step_start
            
            return {
                'step': self.simulation_step,
                'data_points_generated': data_points,
                'simulation_time': step_duration,
                'current_time': self.current_time
            }
        
        def run_what_if_analysis(self, scenario_name: str, 
                               modifications: Dict[str, Any]) -> Dict[str, Any]:
            """Run what-if analysis with different scenarios"""
            
            print(f"Running what-if analysis: {scenario_name}")
            
            # Store original state
            original_state = {
                'vehicles': [(v.vehicle_id, v.battery_level, v.energy_consumption_rate) for v in self.digital_twin.vehicles],
                'stations': [(s.station_id, s.energy_price, s.charging_rate) for s in self.digital_twin.charging_stations]
            }
            
            # Apply modifications
            if 'add_charging_station' in modifications:
                new_station = modifications['add_charging_station']
                self.digital_twin.charging_stations.append(new_station)
            
            if 'change_energy_prices' in modifications:
                price_change = modifications['change_energy_prices']
                for station in self.digital_twin.charging_stations:
                    station.energy_price *= (1 + price_change)
            
            if 'improve_battery_efficiency' in modifications:
                efficiency_improvement = modifications['improve_battery_efficiency']
                for vehicle in self.digital_twin.vehicles:
                    vehicle.energy_consumption_rate *= (1 - efficiency_improvement)
            
            # Run simulation with modifications
            scenario_results = self.run_simulation_step(duration_minutes=30)
            
            # Calculate scenario KPIs
            scenario_kpis = self.calculate_kpis(self.current_time)
            
            # Restore original state
            # (In a real implementation, this would be more sophisticated)
            
            return {
                'scenario': scenario_name,
                'modifications': modifications,
                'results': scenario_results,
                'kpis': scenario_kpis
            }
    
    # Create digital twin orchestrator
    orchestrator = DigitalTwinOrchestrator(digital_twin)
    print(f"Created digital twin orchestrator")
except Exception as e:
    print(f'Error in cell: {e}')
```

### Why This Tier Exists vs Earlier Tiers

The Integrated Digital Twin represents a fundamental shift from optimization to system-wide simulation and monitoring:

**Advantages over Tier 1-4 (All Previous Methods):**
- **Real-time Monitoring**: Continuous synchronization with physical assets vs periodic optimization
-Predictive Analytics**: Forecast future needs and challenges vs reactive problem solving
-System-of-Systems Integration**: Holistic view of entire ecosystem vs isolated route optimization
-What-if Analysis**: Test scenarios and strategies without real-world risks
-Continuous Improvement**: Ongoing learning and adaptation vs one-time solutions

**Unique Capabilities:**
- **IoT Integration**: Real-time sensor data from hundreds of sources
- **Energy Simulation**: Sophisticated modeling of battery dynamics and consumption
- **Environmental Awareness**: Weather, traffic, and grid condition integration
- **Strategic Planning**: Long-term fleet management and infrastructure decisions
- **Performance Monitoring**: Comprehensive KPI tracking and alerting

**Disadvantages vs Earlier Tiers:**
- **Complexity**: Significantly more complex to implement and maintain
- **Infrastructure Requirements**: Needs IoT sensors, communication networks, and computing resources
- **Data Management**: Requires handling massive data streams and storage
- **Initial Investment**: High setup costs for sensors and integration systems

**When to Use This Tier:**
- Large-scale EV fleet operations (50+ vehicles)
- Mission-critical logistics requiring high reliability
- Environments with high operational costs and complexity
- Organizations with digital transformation initiatives
- Long-term strategic fleet management and planning
