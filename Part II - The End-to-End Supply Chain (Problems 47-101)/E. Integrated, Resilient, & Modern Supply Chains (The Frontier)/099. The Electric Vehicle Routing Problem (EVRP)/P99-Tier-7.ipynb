# Problem 99: The Electric Vehicle Routing Problem (EVRP)

## Tier 7: Human-AI Symbiotic Partnership

### Key Assumptions
- Human operators and AI systems collaborate as equal partners in routing decisions
- AI provides explainable recommendations with confidence levels and alternative options
- Humans contribute domain expertise, intuition, and contextual understanding
- Trust and transparency are fundamental to the partnership effectiveness
- Learning flows bidirectionally between human and AI partners

### Approach (Step-by-Step)

The Human-AI Symbiotic Partnership creates a collaborative decision-making environment:

1. **Explainable AI Engine**: Develop AI that can explain its reasoning and decisions
2. **Human Interface Design**: Create intuitive interfaces for human-AI interaction
3. **Trust Building Framework**: Implement mechanisms to build and maintain human trust
4. **Bidirectional Learning**: Enable AI to learn from human feedback and humans to learn from AI
5. **Contextual Awareness**: AI considers human preferences, constraints, and expertise
6. **Collaborative Decision Making**: Joint problem-solving with shared responsibility
7. **Performance Monitoring**: Track partnership effectiveness and improvement over time
8. **Adaptation Mechanisms**: System adapts to individual human partners and their styles

### What to Look for in the Results
- Explainable AI recommendations with clear reasoning
- Human-AI collaboration metrics and trust indicators
- Decision quality improvement through partnership
- Learning and adaptation patterns over time
- Comparison with fully automated and fully manual approaches

### Concrete Example

We will build a human-AI partnership for EV routing:
- 1 depot
- 6 customers with complex constraints
- 2 charging stations with dynamic conditions
- 2 electric vehicles
- Battery capacity: 100 kWh
- Energy consumption: 0.8 kWh per km
- Human-AI collaborative decision making with explainable AI

```python
# Import required libraries for Human-AI Partnership
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass, field
from typing import List, Tuple, Dict, Optional, Any
import random
import time
from datetime import datetime, timedelta
from collections import defaultdict, deque
from enum import Enum
import warnings
warnings.filterwarnings('ignore')

# Set style for professional visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Set random seeds for reproducibility
np.random.seed(42)
random.seed(42)

class DecisionType(Enum):
    ROUTE_PLANNING = "route_planning"
    CHARGING_DECISION = "charging_decision"
    PRIORITY_ADJUSTMENT = "priority_adjustment"
    CONSTRAINT_HANDLING = "constraint_handling"
    EMERGENCY_RESPONSE = "emergency_response"

class ConfidenceLevel(Enum):
    VERY_LOW = 0.2
    LOW = 0.4
    MEDIUM = 0.6
    HIGH = 0.8
    VERY_HIGH = 1.0

@dataclass
class AIRecommendation:
    """AI recommendation with explanation"""
    recommendation_type: DecisionType
    primary_action: str
    confidence: ConfidenceLevel
    reasoning: List[str]
    alternatives: List[str]
    expected_outcome: Dict[str, float]
    risk_factors: List[str]
    data_sources: List[str]
    timestamp: datetime = field(default_factory=datetime.now)
    
@dataclass
class HumanFeedback:
    """Human feedback on AI recommendation"""
    recommendation_id: str
    acceptance_level: float  # 0-1
    modifications: List[str]
    reasoning: str
    confidence_in_decision: float  # 0-1
    additional_context: Dict[str, Any]
    timestamp: datetime = field(default_factory=datetime.now)
    
@dataclass
class TrustMetrics:
    """Trust metrics between human and AI"""
    overall_trust: float  # 0-1
    competence_trust: float  # 0-1
    integrity_trust: float  # 0-1
    benevolence_trust: float  # 0-1
    predictability_trust: float  # 0-1
    trust_history: List[float] = field(default_factory=list)
    
@dataclass
class HumanPartner:
    """Human partner in the symbiotic relationship"""
    partner_id: int
    name: str
    expertise_level: float  # 0-1
    risk_tolerance: float  # 0-1
    decision_style: str  # "analytical", "intuitive", "collaborative"
    preferred_explanation_detail: str  # "brief", "detailed", "technical"
    trust_in_ai: TrustMetrics
    learning_rate: float = 0.1
    adaptation_history: List[Dict[str, Any]] = field(default_factory=list)
    
@dataclass
class CollaborationSession:
    """Record of a human-AI collaboration session"""
    session_id: str
    human_partner: HumanPartner
    start_time: datetime
    end_time: Optional[datetime]
    decisions_made: List[Dict[str, Any]]
    ai_recommendations: List[AIRecommendation]
    human_feedback: List[HumanFeedback]
    collaboration_score: float  # 0-1
    learning_outcomes: List[str]
```

```python
class ExplainableAIEngine:
    """AI engine that provides explainable recommendations"""
    
    def __init__(self):
        self.knowledge_base = {
            'routing_principles': [
                "Minimize total distance while respecting battery constraints",
                "Consider charging station availability and wait times",
                "Prioritize time-sensitive deliveries",
                "Balance workload across vehicles"
            ],
            'battery_management': [
                "Maintain 20% battery buffer for emergencies",
                "Charge during off-peak hours when possible",
                "Consider temperature effects on battery performance",
                "Plan charging stops to minimize total trip time",
            ],
            'operational_constraints': [
                "Respect driver working hours and breaks",
                "Consider customer preferred delivery windows",
                "Account for traffic patterns and road conditions",
                "Follow safety regulations and protocols",
            ]
        }
        
        self.decision_history = deque(maxlen=1000)
        self.explanation_templates = {
            'route_optimization': [
                "Based on current traffic and battery levels, I recommend",
                "Considering the delivery deadlines and charging constraints,",
                "After analyzing the route efficiency and energy requirements,"
            ],
            'charging_decision': [
                "Given the current battery level and upcoming route demands,",
                "Factoring in charging station availability and wait times,",
                "Based on energy consumption predictions and charging costs,",
            ]
        }
        
        self.confidence_calibration = {
            'high_confidence_threshold': 0.8,
            'medium_confidence_threshold': 0.6,
            'low_confidence_threshold': 0.4
        }
    
    def generate_route_recommendation(self, current_state: Dict[str, Any], 
                                      human_preferences: Dict[str, Any]) -> AIRecommendation:
        """Generate explainable route recommendation"""
        
        # Analyze current situation
        battery_level = current_state.get('battery_level', 50.0)
        current_location = current_state.get('location', (0, 0))
        pending_deliveries = current_state.get('pending_deliveries', [])
        
        # Generate primary recommendation
        if battery_level < 30.0:
            primary_action = "Proceed to nearest charging station immediately"
            confidence = ConfidenceLevel.VERY_HIGH
            reasoning = [
                f"Battery level is critically low at {battery_level:.1f}%",
                "Risk of running out of power before completing deliveries",
                "Nearest charging station is available with minimal wait time",
                "Charging now ensures reliable completion of remaining deliveries"
            ]
            alternatives = [
                "Continue to next delivery if very close and urgent",
                "Wait for mobile charging service if available"
            ]
        
        elif len(pending_deliveries) > 3:
            primary_action = "Optimize route for maximum deliveries with one charging stop"
            confidence = ConfidenceLevel.HIGH
            reasoning = [
                f"{len(pending_deliveries)} deliveries pending, requiring efficient routing",
                "Analysis shows optimal route with strategic charging midway",
                "This approach minimizes total distance while meeting all deadlines",
                "Current battery level supports this plan with 20% safety buffer"
            ]
            alternatives = [
                "Split deliveries into two separate routes",
                "Prioritize urgent deliveries only",
                "Add extra charging stop for increased safety margin"
            ]
        
        else:
            primary_action = "Complete remaining deliveries in efficiency order"
            confidence = ConfidenceLevel.MEDIUM
            reasoning = [
                f"Battery level ({battery_level:.1f}%) sufficient for remaining deliveries",
                "Route optimization shows efficient completion without additional charging",
                "Current traffic conditions support this recommendation",
                "All delivery deadlines can be met with this approach"
            ]
            alternatives = [
                "Add preventive charging stop for increased safety",
                "Reorder deliveries based on human preference",
                "Combine with nearby pickup opportunities"
            ]
        
        # Calculate expected outcomes
        expected_outcome = {
            'total_distance': np.random.uniform(15, 25),
            'energy_consumed': np.random.uniform(12, 20),
            'completion_time': np.random.uniform(2, 4),
            'success_probability': confidence.value
        }
        
        # Identify risk factors
        risk_factors = []
        if battery_level < 40.0:
            risk_factors.append("Low battery level increases risk of power depletion")
        if len(pending_deliveries) > 5:
            risk_factors.append("High delivery volume may cause delays")
        if confidence == ConfidenceLevel.MEDIUM:
            risk_factors.append("Moderate confidence due to uncertain factors")
        
        # Data sources
        data_sources = [
            "Real-time battery monitoring system",
            "GPS and traffic data",
            "Delivery priority database",
            "Charging station status API"
        ]
        
        return AIRecommendation(
            recommendation_type=DecisionType.ROUTE_PLANNING,
            primary_action=primary_action,
            confidence=confidence,
            reasoning=reasoning,
            alternatives=alternatives,
            expected_outcome=expected_outcome,
            risk_factors=risk_factors,
            data_sources=data_sources
        )
    
    def generate_charging_recommendation(self, current_state: Dict[str, Any], 
                                        human_preferences: Dict[str, Any]) -> AIRecommendation:
        """Generate explainable charging recommendation"""
        
        battery_level = current_state.get('battery_level', 50.0)
        nearby_stations = current_state.get('nearby_stations', [])
        urgent_deliveries = current_state.get('urgent_deliveries', [])
        
        if battery_level < 25.0:
            primary_action = "Immediate charging required at nearest available station"
            confidence = ConfidenceLevel.VERY_HIGH
            reasoning = [
                f"Critical battery level at {battery_level:.1f}%",
                "Risk of vehicle immobilization is high",
                "Nearest station has available charging point",
                "Charging now prevents service disruption"
            ]
        
        elif battery_level < 45.0 and not urgent_deliveries:
            primary_action = "Opportunistic charging during low-demand period"
            confidence = ConfidenceLevel.HIGH
            reasoning = [
                f"Battery level ({battery_level:.1f}%) allows strategic charging",
                "No urgent deliveries requiring immediate attention",
                "Current time offers lower electricity rates",
                "Proactive charging prevents future constraints"
            ]
        
        else:
            primary_action = "Continue operations, monitor battery level"
            confidence = ConfidenceLevel.MEDIUM
            reasoning = [
                f"Battery level ({battery_level:.1f}%) sufficient for current operations",
                "Charging now would reduce operational efficiency",
                "Better to charge when battery is lower or during breaks",
                "Continuous monitoring ensures optimal charging timing"
            ]
        
        alternatives = [
            "Wait for preferred charging station with faster charging rate",
            "Combine charging with driver break time",
            "Consider mobile charging option if available"
        ]
        
        expected_outcome = {
            'charging_time': np.random.uniform(0.5, 2.0),
            'cost': np.random.uniform(5, 15),
            'battery_after_charging': min(100, battery_level + np.random.uniform(40, 60)),
            'operational_impact': np.random.uniform(0.1, 0.3)
        }
        
        risk_factors = []
        if len(nearby_stations) == 0:
            risk_factors.append("No nearby charging stations available")
        if urgent_deliveries:
            risk_factors.append("Urgent deliveries may be delayed by charging")
        
        data_sources = [
            "Battery management system",
            "Charging station availability API",
            "Electricity pricing data",
            "Delivery schedule database"
        ]
        
        return AIRecommendation(
            recommendation_type=DecisionType.CHARGING_DECISION,
            primary_action=primary_action,
            confidence=confidence,
            reasoning=reasoning,
            alternatives=alternatives,
            expected_outcome=expected_outcome,
            risk_factors=risk_factors,
            data_sources=data_sources
        )
    
    def adapt_to_human_feedback(self, feedback: HumanFeedback):
        """Adapt AI behavior based on human feedback"""
        
        # Update decision patterns based on feedback
        if feedback.acceptance_level < 0.5:
            # Human rejected recommendation, adjust future recommendations
            self.decision_history.append({
                'type': 'rejected',
                'reasoning': feedback.reasoning,
                'modifications': feedback.modifications,
                'timestamp': feedback.timestamp
            })
        
        elif feedback.acceptance_level > 0.8:
            # Human accepted recommendation, reinforce this pattern
            self.decision_history.append({
                'type': 'accepted',
                'reasoning': feedback.reasoning,
                'confidence': feedback.confidence_in_decision,
                'timestamp': feedback.timestamp
            })

# Create explainable AI engine
ai_engine = ExplainableAIEngine()
print(f"Created explainable AI engine with {len(ai_engine.knowledge_base)} knowledge domains")
```

```python
class HumanAIPartnership:
    """Manages the symbiotic partnership between human and AI"""
    
    def __init__(self, human_partner: HumanPartner, ai_engine: ExplainableAIEngine):
        self.human_partner = human_partner
        self.ai_engine = ai_engine
        
        # Partnership state
        self.current_session = None
        self.session_history = []
        
        # Collaboration metrics
        self.collaboration_efficiency = 0.5  # 0-1
        self.decision_quality_score = 0.5  # 0-1
        self.trust_development_rate = 0.0  # Change per session
        
        # Learning and adaptation
        self.shared_knowledge_base = defaultdict(list)
        self.communication_patterns = defaultdict(int)
        
        # Interface preferences
        self.explanation_style = human_partner.preferred_explanation_detail
        self.decision_support_level = "collaborative"  # "advisory", "collaborative", "autonomous"
    
    def start_collaboration_session(self, problem_context: Dict[str, Any]) -> CollaborationSession:
        """Start a new collaboration session"""
        
        session_id = f"session_{datetime.now().strftime('%Y%m%d_%H%M%S')}"
        
        session = CollaborationSession(
            session_id=session_id,
            human_partner=self.human_partner,
            start_time=datetime.now(),
            end_time=None,
            decisions_made=[],
            ai_recommendations=[],
            human_feedback=[],
            collaboration_score=0.0,
            learning_outcomes=[]
        )
        
        self.current_session = session
        return session
    
    def generate_ai_recommendation(self, decision_type: DecisionType, 
                                   current_state: Dict[str, Any]) -> AIRecommendation:
        """Generate AI recommendation adapted to human partner"""
        
        # Get human preferences
        human_preferences = {
            'risk_tolerance': self.human_partner.risk_tolerance,
            'expertise_level': self.human_partner.expertise_level,
            'decision_style': self.human_partner.decision_style,
            'explanation_detail': self.human_partner.preferred_explanation_detail
        }
        
        # Generate base recommendation
        if decision_type == DecisionType.ROUTE_PLANNING:
            recommendation = self.ai_engine.generate_route_recommendation(current_state, human_preferences)
        elif decision_type == DecisionType.CHARGING_DECISION:
            recommendation = self.ai_engine.generate_charging_recommendation(current_state, human_preferences)
        else:
            # Default recommendation for other decision types
            recommendation = AIRecommendation(
                recommendation_type=decision_type,
                primary_action="Consult with human partner for decision",
                confidence=ConfidenceLevel.MEDIUM,
                reasoning=["Complex decision requiring human expertise"],
                alternatives=["Proceed with caution", "Seek additional information"],
                expected_outcome={'success_probability': 0.6},
                risk_factors=["Uncertainty in current situation"],
                data_sources=["Human expertise", "AI analysis"]
            )
        
        # Adapt explanation to human preferences
        recommendation = self._adapt_explanation_to_human(recommendation)
        
        # Adjust confidence based on human trust
        trust_adjustment = self.human_partner.trust_in_ai.overall_trust
        if trust_adjustment < 0.5:
            # Lower confidence if human does not trust AI
            confidence_value = recommendation.confidence.value * (0.5 + trust_adjustment)
            if confidence_value < 0.2:
                recommendation.confidence = ConfidenceLevel.VERY_LOW
            elif confidence_value < 0.4:
                recommendation.confidence = ConfidenceLevel.LOW
            elif confidence_value < 0.6:
                recommendation.confidence = ConfidenceLevel.MEDIUM
            elif confidence_value < 0.8:
                recommendation.confidence = ConfidenceLevel.HIGH
            else:
                recommendation.confidence = ConfidenceLevel.VERY_HIGH
        
        return recommendation
    
    def _adapt_explanation_to_human(self, recommendation: AIRecommendation) -> AIRecommendation:
        """Adapt explanation style to human partner preferences"""
        
        if self.human_partner.preferred_explanation_detail == "brief":
            # Keep only the most important reasoning points
            recommendation.reasoning = recommendation.reasoning[:2]
            recommendation.alternatives = recommendation.alternatives[:1]
        
        elif self.human_partner.preferred_explanation_detail == "detailed":
            # Add additional context and details
            additional_reasoning = [
                "This recommendation is based on current operational data",
                "Multiple factors have been weighted according to your preferences",
                "The expected outcome considers historical performance data"
            ]
            recommendation.reasoning.extend(additional_reasoning)
        
        elif self.human_partner.preferred_explanation_detail == "technical":
            # Add technical details
            technical_reasoning = [
                f"Algorithm confidence: {recommendation.confidence.value:.2f}",
                f"Data freshness: Real-time (last updated: {recommendation.timestamp})",
                "Model version: 2.1 with updated parameters"
            ]
            recommendation.reasoning.extend(technical_reasoning)
        
        return recommendation
    
    def process_human_feedback(self, recommendation: AIRecommendation, 
                               human_response: Dict[str, Any]) -> HumanFeedback:
        """Process and analyze human feedback"""
        
        # Extract feedback components
        acceptance_level = human_response.get('acceptance_level', 0.5)
        modifications = human_response.get('modifications', [])
        reasoning = human_response.get('reasoning', '')
        confidence_in_decision = human_response.get('confidence_in_decision', 0.5)
        additional_context = human_response.get('additional_context', {})
        
        # Create feedback object
        feedback = HumanFeedback(
            recommendation_id=f"rec_{recommendation.timestamp.strftime('%Y%m%d%H%M%S')}",
            acceptance_level=acceptance_level,
            modifications=modifications,
            reasoning=reasoning,
            confidence_in_decision=confidence_in_decision,
            additional_context=additional_context
        )
        
        # Update AI based on feedback
        self.ai_engine.adapt_to_human_feedback(feedback)
        
        # Update trust metrics
        self._update_trust_metrics(feedback)
        
        # Update collaboration patterns
        self._update_collaboration_patterns(feedback)
        
        # Store in current session
        if self.current_session:
            self.current_session.human_feedback.append(feedback)
        
        return feedback
    
    def _update_trust_metrics(self, feedback: HumanFeedback):
        """Update trust metrics based on feedback"""
        
        # Update overall trust
        trust_change = (feedback.acceptance_level - 0.5) * 0.1  # Small incremental changes
        self.human_partner.trust_in_ai.overall_trust = np.clip(
            self.human_partner.trust_in_ai.overall_trust + trust_change, 0.0, 1.0
        )
        
        # Update competence trust based on confidence alignment
        confidence_alignment = abs(feedback.acceptance_level - feedback.confidence_in_decision)
        competence_change = (1 - confidence_alignment) * 0.05
        self.human_partner.trust_in_ai.competence_trust = np.clip(
            self.human_partner.trust_in_ai.competence_trust + competence_change, 0.0, 1.0
        )
        
        # Update integrity trust based on reasoning quality
        if feedback.reasoning and len(feedback.reasoning) > 10:
            integrity_change = 0.02  # Human provided detailed reasoning
        else:
            integrity_change = -0.01  # Limited reasoning
        
        self.human_partner.trust_in_ai.integrity_trust = np.clip(
            self.human_partner.trust_in_ai.integrity_trust + integrity_change, 0.0, 1.0
        )
        
        # Update benevolence trust based on human confidence
        benevolence_change = (feedback.confidence_in_decision - 0.5) * 0.05
        self.human_partner.trust_in_ai.benevolence_trust = np.clip(
            self.human_partner.trust_in_ai.benevolence_trust + benevolence_change, 0.0, 1.0
        )
        
        # Update predictability trust based on consistency
        if self.current_session and len(self.current_session.human_feedback) > 1:
            recent_feedback = self.current_session.human_feedback[-5:]  # Last 5 feedbacks
            acceptance_variance = np.var([f.acceptance_level for f in recent_feedback])
            predictability_change = (1 - acceptance_variance) * 0.03
            self.human_partner.trust_in_ai.predictability_trust = np.clip(
                self.human_partner.trust_in_ai.predictability_trust + predictability_change, 0.0, 1.0
            )
        
        # Store trust history
        self.human_partner.trust_in_ai.trust_history.append(self.human_partner.trust_in_ai.overall_trust)
    
    def _update_collaboration_patterns(self, feedback: HumanFeedback):
        """Update collaboration patterns based on feedback"""
        
        # Track communication patterns
        if feedback.acceptance_level > 0.7:
            self.communication_patterns['high_acceptance'] += 1
        elif feedback.acceptance_level < 0.3:
            self.communication_patterns['low_acceptance'] += 1
        
        if feedback.modifications:
            self.communication_patterns['human_modifications'] += 1
        
        # Update collaboration efficiency
        if self.current_session and len(self.current_session.human_feedback) > 0:
            recent_acceptance = np.mean([f.acceptance_level for f in self.current_session.human_feedback])
            self.collaboration_efficiency = 0.7 * self.collaboration_efficiency + 0.3 * recent_acceptance
    
    def calculate_collaboration_score(self) -> float:
        """Calculate overall collaboration score"""
        
        if not self.current_session:
            return 0.0
        
        # Factors affecting collaboration score
        trust_score = self.human_partner.trust_in_ai.overall_trust
        efficiency_score = self.collaboration_efficiency
        
        # Decision quality (based on human confidence and acceptance)
        if self.current_session.human_feedback:
            avg_acceptance = np.mean([f.acceptance_level for f in self.current_session.human_feedback])
            avg_confidence = np.mean([f.confidence_in_decision for f in self.current_session.human_feedback])
            decision_quality = (avg_acceptance + avg_confidence) / 2
        else:
            decision_quality = 0.5
        
        # Learning outcomes
        learning_score = len(self.current_session.learning_outcomes) / 10.0  # Normalize by expected learning
        learning_score = min(learning_score, 1.0)
        
        # Weighted combination
        collaboration_score = (
            0.3 * trust_score +
            0.3 * efficiency_score +
            0.25 * decision_quality +
            0.15 * learning_score
        )
        
        return collaboration_score

# Create human partner
human_partner = HumanPartner(
    partner_id=1,
    name="Alex Chen",
    expertise_level=0.8,
    risk_tolerance=0.4,
    decision_style="analytical",
    preferred_explanation_detail="detailed",
    trust_in_ai=TrustMetrics(
        overall_trust=0.6,
        competence_trust=0.7,
        integrity_trust=0.8,
        benevolence_trust=0.6,
        predictability_trust=0.5
    )
)

# Create partnership
partnership = HumanAIPartnership(human_partner, ai_engine)
print(f"Created human-AI partnership with {human_partner.name}")
print(f"Initial trust level: {human_partner.trust_in_ai.overall_trust:.2f}")
print(f"Decision style: {human_partner.decision_style}")
print(f"Preferred explanation: {human_partner.preferred_explanation_detail}")
```

```python
def simulate_human_ai_collaboration():
    """Simulate a complete human-AI collaboration session"""
    
    print("Starting Human-AI Collaboration Session")
    print("=" * 50)
    
    # Initialize session
    problem_context = {
        'problem_type': 'ev_routing',
        'urgency': 'medium',
        'constraints': ['battery_limit', 'time_windows', 'charging_availability']
    }
    
    session = partnership.start_collaboration_session(problem_context)
    print(f"Session {session.session_id} started at {session.start_time.strftime('%H:%M:%S')}")
    
    # Simulate multiple decision points
    decision_scenarios = [
        {
            'type': DecisionType.ROUTE_PLANNING,
            'state': {
                'battery_level': 35.0,
                'location': (10, 15),
                'pending_deliveries': [
                    {'id': 1, 'location': (20, 25), 'deadline': '14:00', 'priority': 'high'},
                    {'id': 2, 'location': (8, 30), 'deadline': '15:30', 'priority': 'medium'},
                    {'id': 3, 'location': (25, 10), 'deadline': '16:00', 'priority': 'low'}
                ],
                'nearby_stations': [(12, 18), (22, 12)]
            }
        },
        {
            'type': DecisionType.CHARGING_DECISION,
            'state': {
                'battery_level': 28.0,
                'location': (20, 25),
                'urgent_deliveries': [{'id': 1, 'deadline': '13:30'}],
                'nearby_stations': [(22, 22), (18, 28)]
            }
        },
        {
            'type': DecisionType.ROUTE_PLANNING,
            'state': {
                'battery_level': 75.0,
                'location': (15, 20),
                'pending_deliveries': [
                    {'id': 4, 'location': (30, 15), 'deadline': '17:00', 'priority': 'medium'},
                    {'id': 5, 'location': (10, 8), 'deadline': '18:00', 'priority': 'low'}
                ],
                'nearby_stations': [(18, 18), (25, 25)]
            }
        }
    ]
    
    # Process each decision scenario
    for i, scenario in enumerate(decision_scenarios):
        print(f"\nDecision Point {i+1}: {scenario['type'].value}")
        print("-" * 30)
        
        # Generate AI recommendation
        recommendation = partnership.generate_ai_recommendation(
            scenario['type'], scenario['state']
        )
        
        # Display AI recommendation
        print(f"\nAI Recommendation (Confidence: {recommendation.confidence.value:.2f}):")
        print(f"Primary Action: {recommendation.primary_action}")
        print(f"\nReasoning:")
        for j, reason in enumerate(recommendation.reasoning, 1):
            print(f"  {j}. {reason}")
        
        if recommendation.alternatives:
            print(f"\nAlternatives:")
            for j, alt in enumerate(recommendation.alternatives, 1):
                print(f"  {j}. {alt}")
        
        if recommendation.risk_factors:
            print(f"\nRisk Factors:")
            for j, risk in enumerate(recommendation.risk_factors, 1):
                print(f"  {j}. {risk}")
        
        print(f"\nExpected Outcomes:")
        for key, value in recommendation.expected_outcome.items():
            print(f"  {key}: {value:.2f}")
        
        # Simulate human response
        human_response = simulate_human_response(recommendation, human_partner)
        
        print(f"\nHuman Response:")
        print(f"Acceptance Level: {human_response['acceptance_level']:.2f}")
        print(f"Reasoning: {human_response['reasoning']}")
        print(f"Confidence in Decision: {human_response['confidence_in_decision']:.2f}")
        
        if human_response['modifications']:
            print(f"Modifications: {', '.join(human_response['modifications'])}")
        
        # Process feedback
        feedback = partnership.process_human_feedback(recommendation, human_response)
        
        # Store in session
        session.ai_recommendations.append(recommendation)
        session.decisions_made.append({
            'scenario': scenario,
            'recommendation': recommendation,
            'human_response': human_response,
            'final_decision': human_response.get('final_action', recommendation.primary_action)
        })
    
    # End session
    session.end_time = datetime.now()
    session.collaboration_score = partnership.calculate_collaboration_score()
    
    # Generate learning outcomes
    session.learning_outcomes = generate_learning_outcomes(session)
    
    partnership.session_history.append(session)
    
    print(f"\nSession completed at {session.end_time.strftime('%H:%M:%S')}")
    print(f"Collaboration Score: {session.collaboration_score:.2f}")
    print(f"Learning Outcomes: {len(session.learning_outcomes)}")
    
    return session

def simulate_human_response(recommendation: AIRecommendation, human_partner: HumanPartner) -> Dict[str, Any]:
    """Simulate human response to AI recommendation"""
    
    # Base acceptance probability based on trust and confidence
    base_acceptance = (human_partner.trust_in_ai.overall_trust + recommendation.confidence.value) / 2
    
    # Adjust based on human characteristics
    if human_partner.decision_style == "analytical":
        # Analytical humans are more likely to accept high-confidence recommendations
        if recommendation.confidence.value > 0.7:
            base_acceptance += 0.2
        else:
            base_acceptance -= 0.1
    
    elif human_partner.decision_style == "intuitive":
        # Intuitive humans may be more skeptical of AI recommendations
        base_acceptance -= 0.1
    
    # Add randomness
    acceptance_level = np.clip(np.random.normal(base_acceptance, 0.2), 0.0, 1.0)
    
    # Generate response based on acceptance level
    if acceptance_level > 0.7:
        reasoning = "AI recommendation aligns with my analysis and current situation"
        modifications = []
        confidence_in_decision = 0.8 + np.random.uniform(0, 0.2)
        final_action = recommendation.primary_action
    
    elif acceptance_level > 0.4:
        reasoning = "AI recommendation is reasonable but I would like to modify it slightly"
        modifications = ["Adjust timing", "Consider additional factors"]
        confidence_in_decision = 0.6 + np.random.uniform(-0.1, 0.2)
        final_action = recommendation.alternatives[0] if recommendation.alternatives else recommendation.primary_action
    
    else:
        reasoning = "I disagree with the AI recommendation based on my experience"
        modifications = ["Change approach completely", "Consider different priorities"]
        confidence_in_decision = 0.7 + np.random.uniform(0, 0.3)
        final_action = "Manual decision based on human judgment"
    
    return {
        'acceptance_level': acceptance_level,
        'modifications': modifications,
        'reasoning': reasoning,
        'confidence_in_decision': confidence_in_decision,
        'additional_context': {
            'human_expertise_applied': human_partner.expertise_level,
            'time_pressure': np.random.uniform(0, 1)
        },
        'final_action': final_action
    }

def generate_learning_outcomes(session: CollaborationSession) -> List[str]:
    """Generate learning outcomes from the session"""
    
    outcomes = []
    
    # Analyze feedback patterns
    if session.human_feedback:
        avg_acceptance = np.mean([f.acceptance_level for f in session.human_feedback])
        
        if avg_acceptance > 0.7:
            outcomes.append("High alignment between AI and human preferences")
        elif avg_acceptance < 0.4:
            outcomes.append("AI recommendations need better adaptation to human preferences")
        else:
            outcomes.append("Moderate alignment with room for improvement")
    
    # Analyze trust development
    if len(session.human_partner.trust_in_ai.trust_history) > 1:
        trust_change = session.human_partner.trust_in_ai.trust_history[-1] - session.human_partner.trust_in_ai.trust_history[-2]
        if trust_change > 0.05:
            outcomes.append("Trust in AI improved during this session")
        elif trust_change < -0.05:
            outcomes.append("Trust in AI decreased, needs attention")
        else:
            outcomes.append("Trust in AI remained stable")
    
    # Analyze decision quality
    avg_confidence = np.mean([f.confidence_in_decision for f in session.human_feedback])
    if avg_confidence > 0.7:
        outcomes.append("High confidence in collaborative decisions")
    else:
        outcomes.append("Need to improve decision confidence through better explanations")
    
    return outcomes

# Run the collaboration simulation
session = simulate_human_ai_collaboration()
```

```python
def analyze_partnership_evolution(partnership: HumanAIPartnership, num_sessions: int = 5):
    """Analyze partnership evolution over multiple sessions"""
    
    print("Analyzing Partnership Evolution")
    print("=" * 50)
    
    # Track metrics over sessions
    trust_evolution = []
    collaboration_scores = []
    acceptance_rates = []
    
    # Simulate multiple sessions
    for session_num in range(num_sessions):
        print(f"\nSession {session_num + 1}/{num_sessions}")
        print("-" * 30)
        
        # Simulate session with evolving scenarios
        session = simulate_human_ai_collaboration()
        
        # Track metrics
        trust_evolution.append(partnership.human_partner.trust_in_ai.overall_trust)
        collaboration_scores.append(session.collaboration_score)
        
        if session.human_feedback:
            avg_acceptance = np.mean([f.acceptance_level for f in session.human_feedback])
            acceptance_rates.append(avg_acceptance)
        else:
            acceptance_rates.append(0.5)
        
        # Show key metrics
        print(f"Trust Level: {partnership.human_partner.trust_in_ai.overall_trust:.3f}")
        print(f"Collaboration Score: {session.collaboration_score:.3f}")
        print(f"Acceptance Rate: {acceptance_rates[-1]:.3f}")
        
        # Adapt human partner slightly over time
        if session_num > 2:
            # Human becomes more comfortable with AI
            partnership.human_partner.trust_in_ai.overall_trust += 0.02
            partnership.human_partner.expertise_level = min(1.0, partnership.human_partner.expertise_level + 0.01)
    
    # Final analysis
    print(f"\nPartnership Evolution Analysis:")
    print("=" * 50)
    
    # Trust evolution
    trust_trend = np.polyfit(range(len(trust_evolution)), trust_evolution, 1)[0]
    print(f"Trust Trend: {'Improving' if trust_trend > 0 else 'Declining' if trust_trend < 0 else 'Stable'}")
    print(f"Initial Trust: {trust_evolution[0]:.3f}")
    print(f"Final Trust: {trust_evolution[-1]:.3f}")
    print(f"Trust Change: {trust_evolution[-1] - trust_evolution[0]:+.3f}")
    
    # Collaboration improvement
    collab_trend = np.polyfit(range(len(collaboration_scores)), collaboration_scores, 1)[0]
    print(f"\nCollaboration Trend: {'Improving' if collab_trend > 0 else 'Declining' if collab_trend < 0 else 'Stable'}")
    print(f"Average Collaboration Score: {np.mean(collaboration_scores):.3f}")
    print(f"Best Collaboration Score: {np.max(collaboration_scores):.3f}")
    
    # Acceptance patterns
    avg_acceptance = np.mean(acceptance_rates)
    acceptance_std = np.std(acceptance_rates)
    print(f"\nAcceptance Analysis:")
    print(f"Average Acceptance Rate: {avg_acceptance:.3f}")
    print(f"Acceptance Consistency: {'High' if acceptance_std < 0.2 else 'Medium' if acceptance_std < 0.4 else 'Low'}")
    
    # Trust component analysis
    trust_components = {
        'Competence': partnership.human_partner.trust_in_ai.competence_trust,
        'Integrity': partnership.human_partner.trust_in_ai.integrity_trust,
        'Benevolence': partnership.human_partner.trust_in_ai.benevolence_trust,
        'Predictability': partnership.human_partner.trust_in_ai.predictability_trust
    }
    
    print(f"\nTrust Component Analysis:")
    for component, value in trust_components.items():
        status = "Strong" if value > 0.7 else "Moderate" if value > 0.5 else "Needs Improvement"
        print(f"  {component}: {value:.3f} ({status})")
    
    return {
        'trust_evolution': trust_evolution,
        'collaboration_scores': collaboration_scores,
        'acceptance_rates': acceptance_rates,
        'trust_trend': trust_trend,
        'collab_trend': collab_trend,
        'trust_components': trust_components
    }

# Analyze partnership evolution
evolution_analysis = analyze_partnership_evolution(partnership, num_sessions=5)
```

```python
def visualize_human_ai_partnership(partnership: HumanAIPartnership, evolution_analysis: Dict):
    """Create comprehensive visualization of the human-AI partnership"""
    
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('Electric Vehicle Routing Problem - Human-AI Symbiotic Partnership', fontsize=16, fontweight='bold')
    
    # Plot 1: Trust evolution over sessions
    sessions = range(1, len(evolution_analysis['trust_evolution']) + 1)
    ax1.plot(sessions, evolution_analysis['trust_evolution'], 'b-', linewidth=2, marker='o', markersize=6)
    ax1.set_xlabel('Session Number')
    ax1.set_ylabel('Trust Level')
    ax1.set_title('Trust Evolution Over Time')
    ax1.set_ylim(0, 1)
    ax1.grid(True, alpha=0.3)
    
    # Add trend line
    z = np.polyfit(sessions, evolution_analysis['trust_evolution'], 1)
    p = np.poly1d(z)
    ax1.plot(sessions, p(sessions), "r--", alpha=0.8, label=f'Trend: {z[0]:.3f}')
    ax1.legend()
    
    # Plot 2: Collaboration scores and acceptance rates
    ax2_twin = ax2.twinx()
    
    line1 = ax2.plot(sessions, evolution_analysis['collaboration_scores'], 'g-', linewidth=2, marker='s', markersize=6, label='Collaboration Score')
    line2 = ax2_twin.plot(sessions, evolution_analysis['acceptance_rates'], 'orange', linewidth=2, marker='^', markersize=6, label='Acceptance Rate')
    
    ax2.set_xlabel('Session Number')
    ax2.set_ylabel('Collaboration Score', color='green')
    ax2_twin.set_ylabel('Acceptance Rate', color='orange')
    ax2.set_title('Partnership Performance Metrics')
    ax2.tick_params(axis='y', labelcolor='green')
    ax2_twin.tick_params(axis='y', labelcolor='orange')
    ax2.grid(True, alpha=0.3)
    
    # Plot 3: Trust component breakdown
    trust_components = evolution_analysis['trust_components']
    components = list(trust_components.keys())
    values = list(trust_components.values())
    
    colors = ['skyblue', 'lightgreen', 'gold', 'lightcoral']
    bars = ax3.bar(components, values, color=colors, alpha=0.8)
    
    ax3.set_ylabel('Trust Level')
    ax3.set_title('Trust Component Analysis')
    ax3.set_ylim(0, 1)
    ax3.grid(True, alpha=0.3)
    
    # Add value labels on bars
    for bar, value in zip(bars, values):
        height = bar.get_height()
        ax3.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                f'{value:.3f}', ha='center', va='bottom')
    
    # Plot 4: Partnership insights and statistics
    human_partner = partnership.human_partner
    
    insights_text = f"""
    Human-AI Partnership Insights:
    ─────────────────────────────
    Partner: {human_partner.name}
    Expertise Level: {human_partner.expertise_level:.2f}
    Decision Style: {human_partner.decision_style.title()}
    Risk Tolerance: {human_partner.risk_tolerance:.2f}
    
    Trust Metrics:
    ─────────────────────────────
    Overall Trust: {human_partner.trust_in_ai.overall_trust:.3f}
    Trust Trend: {'Improving' if evolution_analysis['trust_trend'] > 0 else 'Declining' if evolution_analysis['trust_trend'] < 0 else 'Stable'}
    Trust Change: {evolution_analysis['trust_evolution'][-1] - evolution_analysis['trust_evolution'][0]:+.3f}
    
    Performance Metrics:
    ─────────────────────────────
    Avg Collaboration: {np.mean(evolution_analysis['collaboration_scores']):.3f}
    Best Collaboration: {np.max(evolution_analysis['collaboration_scores']):.3f}
    Avg Acceptance: {np.mean(evolution_analysis['acceptance_rates']):.3f}
    Acceptance Stability: {'High' if np.std(evolution_analysis['acceptance_rates']) < 0.2 else 'Medium'}
    
    Partnership Characteristics:
    ─────────────────────────────
    Explanation Style: {human_partner.preferred_explanation_detail.title()}
    Collaboration Mode: Collaborative
    Sessions Completed: {len(evolution_analysis['trust_evolution'])}
    Learning Rate: {human_partner.learning_rate:.2f}
    
    Key Strengths:
    ─────────────────────────────
    • Explainable AI recommendations
    • Bidirectional learning
    • Trust-based adaptation
    • Context-aware decision support
    
    Training Characteristics:
    ─────────────────────────────
    • Sample efficiency: {len(partnership.ai_engine.decision_history)} experiences
    • Convergence: {'Stable' if len(evolution_analysis['trust_evolution']) > 100 else 'Early stage'}
    • Exploration: {'Balanced' if 0.01 < evolution_analysis['trust_evolution'][-1] < 0.1 else 'Low'}
    """
    
    ax4.text(0.05, 0.95, insights_text, transform=ax4.transAxes, fontsize=9,
             verticalalignment='top', fontfamily='monospace',
             bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))
    ax4.set_xlim(0, 1)
    ax4.set_ylim(0, 1)
    ax4.axis('off')
    ax4.set_title('Partnership Analysis Dashboard', fontweight='bold')
    
    plt.tight_layout()
    plt.show()

# Visualize the partnership
visualize_human_ai_partnership(partnership, evolution_analysis)
```
