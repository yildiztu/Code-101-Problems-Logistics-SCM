# Problem 99: The Electric Vehicle Routing Problem (EVRP)

## Tier 8: Value-Aligned & Ethical Framework

### Key Assumptions
- EV routing decisions must align with multiple stakeholder values and ethical principles
- Environmental sustainability, social equity, and economic viability are equally important
- Ethical constraints are enforceable through constitutional AI frameworks
- Stakeholder preferences are explicitly represented and weighted in decision making
- Transparency and accountability are fundamental requirements for the system

### Approach (Step-by-Step)

The Value-Aligned & Ethical Framework creates EV routing solutions that respect ethical principles:

1. **Stakeholder Analysis**: Identify all relevant stakeholders and their values
2. **Ethical Principles Definition**: Establish core ethical principles and constraints
3. **Value Modeling**: Create mathematical models of stakeholder utilities and preferences
4. **Constitutional AI Design**: Implement constitutional constraints in the optimization
5. **Multi-Objective Optimization**: Balance competing values and ethical requirements
6. **Transparency Mechanisms**: Ensure all decisions are explainable and auditable
7. **Accountability Systems**: Track compliance and ethical performance metrics
8. **Adaptive Ethics**: Learn from stakeholder feedback and evolve ethical frameworks

### What to Look for in the Results
- Multi-stakeholder utility optimization with ethical constraints
- Transparent decision making with clear ethical reasoning
- Balance between environmental, social, and economic objectives
- Accountability metrics and compliance tracking
- Comparison with utilitarian (profit-only) approaches

### Concrete Example

We'll build a value-aligned EV routing system:
- 1 depot
- 6 customers with varying social importance levels
- 2 charging stations with renewable energy availability
- 2 electric vehicles
- Battery capacity: 100 kWh
- Energy consumption: 0.8 kWh per km
- Multi-stakeholder optimization with ethical constraints

```python
# Import required libraries for Value-Aligned & Ethical Framework
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass, field
from typing import List, Tuple, Dict, Optional, Any
import random
import time
from datetime import datetime, timedelta
from collections import defaultdict, deque
from enum import Enum
import warnings
warnings.filterwarnings('ignore')

# Set style for professional visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Set random seeds for reproducibility
np.random.seed(42)
random.seed(42)

class EthicalPrinciple(Enum):
    ENVIRONMENTAL_SUSTAINABILITY = "environmental_sustainability"
    SOCIAL_EQUITY = "social_equity"
    ECONOMIC_VIABILITY = "economic_viability"
    TRANSPARENCY = "transparency"
    ACCOUNTABILITY = "accountability"
    FAIRNESS = "fairness"
    SAFETY = "safety"
    PRIVACY = "privacy"

class StakeholderType(Enum):
    CUSTOMER = "customer"
    ENVIRONMENT = "environment"
    COMMUNITY = "community"
    EMPLOYEE = "employee"
    INVESTOR = "investor"
    REGULATOR = "regulator"
    SOCIETY = "society"

@dataclass
class Stakeholder:
    """Represents a stakeholder in the EV routing system"""
    stakeholder_id: int
    stakeholder_type: StakeholderType
    name: str
    utility_function: str  # "linear", "concave", "convex", "step"
    weight: float  # Importance weight in decision making
    preferences: Dict[str, float] = field(default_factory=dict)
    constraints: List[str] = field(default_factory=list)
    ethical_principles: List[EthicalPrinciple] = field(default_factory=list)
    
@dataclass
class EthicalConstraint:
    """Represents an ethical constraint in the optimization"""
    constraint_id: int
    principle: EthicalPrinciple
    description: str
    constraint_type: str  # "hard", "soft", "penalty"
    threshold: float
    penalty_weight: float = 1.0
    measurement_function: str = "linear"
    
@dataclass
class ValueAlignedSolution:
    """Solution that respects ethical constraints and stakeholder values"""
    routes: List[List[Tuple[float, float]]]
    stakeholder_utilities: Dict[int, float]
    ethical_compliance: Dict[EthicalPrinciple, float]
    total_utility: float
    environmental_impact: float
    social_impact: float
    economic_cost: float
    transparency_score: float
    accountability_metrics: Dict[str, float]
    ethical_reasoning: List[str]
    
@dataclass
class EthicalAudit:
    """Audit trail for ethical compliance"""
    audit_id: str
    timestamp: datetime
    solution: ValueAlignedSolution
    compliance_score: float
    violations: List[str]
    stakeholder_feedback: Dict[int, str]
    recommendations: List[str]
    ethical_score_improvement: float
```

```python
class ConstitutionalAIFramework:
    """Constitutional AI framework for ethical EV routing"""
    
    def __init__(self):
        self.constitution = {
            EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY: {
                "description": "Minimize environmental impact and promote sustainability",
                "constraints": [
                    "Maximize use of renewable energy for charging",
                    "Minimize total energy consumption",
                    "Prefer routes with lower carbon footprint",
                    "Avoid environmentally sensitive areas"
                ],
                "metrics": ["co2_emissions", "energy_efficiency", "renewable_energy_usage"]
            },
            EthicalPrinciple.SOCIAL_EQUITY: {
                "description": "Ensure fair and equitable service to all communities",
                "constraints": [
                    "Provide equal service quality regardless of location",
                    "Prioritize underserved communities",
                    "Ensure fair distribution of environmental burdens",
                    "Maintain accessibility for all customers"
                ],
                "metrics": ["service_equity", "community_impact", "accessibility_score"]
            },
            EthicalPrinciple.ECONOMIC_VIABILITY: {
                "description": "Ensure economic sustainability and efficiency",
                "constraints": [
                    "Maintain profitable operations",
                    "Optimize resource utilization",
                    "Minimize operational costs",
                    "Ensure long-term financial sustainability"
                ],
                "metrics": ["profit_margin", "cost_efficiency", "resource_utilization"]
            },
            EthicalPrinciple.TRANSPARENCY: {
                "description": "Ensure transparent and explainable decision making",
                "constraints": [
                    "Provide clear explanations for routing decisions",
                    "Make stakeholder impact assessments visible",
                    "Document ethical reasoning for all decisions",
                    "Enable audit trails for accountability"
                ],
                "metrics": ["explanation_quality", "audit_completeness", "reasoning_transparency"]
            },
            EthicalPrinciple.ACCOUNTABILITY: {
                "description": "Ensure accountability for ethical performance",
                "constraints": [
                    "Track ethical compliance metrics",
                    "Provide mechanisms for stakeholder feedback",
                    "Enable corrective actions for violations",
                    "Maintain records of ethical decision making"
                ],
                "metrics": ["compliance_rate", "feedback_response", "violation_resolution"]
            }
        }
        
        self.ethical_constraints = []
        self.compliance_history = deque(maxlen=1000)
        self.audit_trail = []
    
    def add_ethical_constraint(self, principle: EthicalPrinciple, description: str, 
                               constraint_type: str, threshold: float, 
                               penalty_weight: float = 1.0):
        """Add an ethical constraint to the framework"""
        
        constraint = EthicalConstraint(
            constraint_id=len(self.ethical_constraints),
            principle=principle,
            description=description,
            constraint_type=constraint_type,
            threshold=threshold,
            penalty_weight=penalty_weight
        )
        
        self.ethical_constraints.append(constraint)
    
    def evaluate_ethical_compliance(self, solution: ValueAlignedSolution) -> Dict[EthicalPrinciple, float]:
        """Evaluate solution compliance with ethical principles"""
        
        compliance_scores = {}
        
        for principle in EthicalPrinciple:
            if principle == EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY:
                # Environmental compliance based on CO2 emissions and renewable energy usage
                co2_score = max(0, 1 - solution.environmental_impact / 100)  # Normalize to 0-1
                renewable_score = solution.accountability_metrics.get('renewable_energy_usage', 0.5)
                compliance_scores[principle] = (co2_score + renewable_score) / 2
            
            elif principle == EthicalPrinciple.SOCIAL_EQUITY:
                # Social equity based on service distribution and community impact
                equity_score = solution.accountability_metrics.get('service_equity', 0.5)
                community_score = solution.social_impact / 100  # Normalize to 0-1
                compliance_scores[principle] = (equity_score + community_score) / 2
            
            elif principle == EthicalPrinciple.ECONOMIC_VIABILITY:
                # Economic viability based on cost efficiency and profit margin
                cost_score = max(0, 1 - solution.economic_cost / 1000)  # Normalize to 0-1
                efficiency_score = solution.accountability_metrics.get('cost_efficiency', 0.5)
                compliance_scores[principle] = (cost_score + efficiency_score) / 2
            
            elif principle == EthicalPrinciple.TRANSPARENCY:
                # Transparency based on explanation quality and reasoning visibility
                explanation_score = len(solution.ethical_reasoning) / 10  # More reasoning = more transparent
                audit_score = solution.accountability_metrics.get('audit_completeness', 0.5)
                compliance_scores[principle] = min(1, (explanation_score + audit_score) / 2)
            
            elif principle == EthicalPrinciple.ACCOUNTABILITY:
                # Accountability based on compliance tracking and feedback mechanisms
                compliance_rate = solution.accountability_metrics.get('compliance_rate', 0.5)
                feedback_score = solution.accountability_metrics.get('feedback_response', 0.5)
                compliance_scores[principle] = (compliance_rate + feedback_score) / 2
            
            else:
                # Default compliance score for other principles
                compliance_scores[principle] = 0.7  # Assume reasonable compliance
        
        return compliance_scores
    
    def calculate_ethical_penalty(self, solution: ValueAlignedSolution) -> float:
        """Calculate penalty for ethical violations"""
        
        compliance_scores = self.evaluate_ethical_compliance(solution)
        total_penalty = 0.0
        
        for constraint in self.ethical_constraints:
            compliance = compliance_scores.get(constraint.principle, 0.5)
            
            if compliance < constraint.threshold:
                if constraint.constraint_type == "hard":
                    # Hard constraints make solution infeasible
                    return float('inf')
                elif constraint.constraint_type == "soft":
                    # Soft constraints add penalty
                    violation = constraint.threshold - compliance
                    penalty = violation * constraint.penalty_weight
                    total_penalty += penalty
                elif constraint.constraint_type == "penalty":
                    # Penalty constraints add weighted penalty
                    penalty = (1 - compliance) * constraint.penalty_weight
                    total_penalty += penalty
        
        return total_penalty
    
    def generate_ethical_reasoning(self, solution: ValueAlignedSolution) -> List[str]:
        """Generate ethical reasoning for the solution"""
        
        reasoning = []
        compliance_scores = self.evaluate_ethical_compliance(solution)
        
        # Environmental reasoning
        env_compliance = compliance_scores.get(EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY, 0.5)
        if env_compliance > 0.8:
            reasoning.append("Route minimizes environmental impact through efficient energy usage")
        elif env_compliance > 0.6:
            reasoning.append("Route balances environmental concerns with operational needs")
        else:
            reasoning.append("Route prioritizes other factors over environmental optimization")
        
        # Social equity reasoning
        social_compliance = compliance_scores.get(EthicalPrinciple.SOCIAL_EQUITY, 0.5)
        if social_compliance > 0.8:
            reasoning.append("Route ensures equitable service to all communities")
        elif social_compliance > 0.6:
            reasoning.append("Route provides reasonable service equity with some trade-offs")
        else:
            reasoning.append("Route requires improvement in social equity considerations")
        
        # Economic viability reasoning
        econ_compliance = compliance_scores.get(EthicalPrinciple.ECONOMIC_VIABILITY, 0.5)
        if econ_compliance > 0.8:
            reasoning.append("Route maintains excellent economic efficiency and sustainability")
        elif econ_compliance > 0.6:
            reasoning.append("Route balances economic efficiency with other ethical considerations")
        else:
            reasoning.append("Route prioritizes ethical goals over economic efficiency")
        
        # Transparency reasoning
        trans_compliance = compliance_scores.get(EthicalPrinciple.TRANSPARENCY, 0.5)
        reasoning.append(f"Decision transparency score: {trans_compliance:.2f}")
        
        # Accountability reasoning
        acc_compliance = compliance_scores.get(EthicalPrinciple.ACCOUNTABILITY, 0.5)
        reasoning.append(f"Accountability measures implemented with {acc_compliance:.2f} effectiveness")
        
        return reasoning

# Create constitutional AI framework
constitutional_ai = ConstitutionalAIFramework()

# Add ethical constraints
constitutional_ai.add_ethical_constraint(
    EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY,
    "CO2 emissions must be below threshold",
    "soft",
    0.7,
    penalty_weight=2.0
)

constitutional_ai.add_ethical_constraint(
    EthicalPrinciple.SOCIAL_EQUITY,
    "Service equity must be maintained",
    "soft",
    0.6,
    penalty_weight=1.5
)

constitutional_ai.add_ethical_constraint(
    EthicalPrinciple.ECONOMIC_VIABILITY,
    "Cost efficiency must be reasonable",
    "soft",
    0.5,
    penalty_weight=1.0
)

print(f"Created Constitutional AI framework with {len(constitutional_ai.ethical_constraints)} ethical constraints")
print(f"Constitution covers {len(constitutional_ai.constitution)} ethical principles")
```

```python
class StakeholderUtilityOptimizer:
    """Optimizes stakeholder utilities while respecting ethical constraints"""
    
    def __init__(self, stakeholders: List[Stakeholder], constitutional_ai: ConstitutionalAIFramework):
        self.stakeholders = stakeholders
        self.constitutional_ai = constitutional_ai
        
        # Stakeholder weights for decision making
        self.stakeholder_weights = {s.stakeholder_id: s.weight for s in stakeholders}
        
        # Utility function definitions
        self.utility_functions = {
            "linear": lambda x, params: params.get('slope', 1) * x + params.get('intercept', 0),
            "concave": lambda x, params: params.get('scale', 1) * np.sqrt(x),
            "convex": lambda x, params: params.get('scale', 1) * x**2,
            "step": lambda x, params: params.get('value', 1) if x >= params.get('threshold', 0.5) else 0
        }
        
        # Optimization history
        self.optimization_history = deque(maxlen=100)
        self.utility_evolution = []
    
    def calculate_stakeholder_utility(self, solution_metrics: Dict[str, float], 
                                     stakeholder: Stakeholder) -> float:
        """Calculate utility for a specific stakeholder"""
        
        utility = 0.0
        utility_func = self.utility_functions.get(stakeholder.utility_function, self.utility_functions["linear"])
        
        # Calculate utility based on stakeholder type and preferences
        if stakeholder.stakeholder_type == StakeholderType.CUSTOMER:
            # Customers value service quality, timeliness, and cost
            service_quality = solution_metrics.get('service_quality', 0.5)
            timeliness = solution_metrics.get('timeliness', 0.5)
            cost_affordability = 1 - solution_metrics.get('cost_factor', 0.5)
            
            utility = utility_func(service_quality, {'slope': stakeholder.preferences.get('service_weight', 0.4)})
            utility += utility_func(timeliness, {'slope': stakeholder.preferences.get('time_weight', 0.3)})
            utility += utility_func(cost_affordability, {'slope': stakeholder.preferences.get('cost_weight', 0.3)})
        
        elif stakeholder.stakeholder_type == StakeholderType.ENVIRONMENT:
            # Environment values low emissions, renewable energy, and efficiency
            low_emissions = 1 - solution_metrics.get('emissions_level', 0.5)
            renewable_usage = solution_metrics.get('renewable_energy', 0.5)
            efficiency = solution_metrics.get('energy_efficiency', 0.5)
            
            utility = utility_func(low_emissions, {'scale': stakeholder.preferences.get('emissions_weight', 0.4)})
            utility += utility_func(renewable_usage, {'scale': stakeholder.preferences.get('renewable_weight', 0.3)})
            utility += utility_func(efficiency, {'scale': stakeholder.preferences.get('efficiency_weight', 0.3)})
        
        elif stakeholder.stakeholder_type == StakeholderType.COMMUNITY:
            # Community values local employment, reduced traffic, and accessibility
            local_jobs = solution_metrics.get('local_employment', 0.5)
            traffic_reduction = solution_metrics.get('traffic_reduction', 0.5)
            accessibility = solution_metrics.get('accessibility', 0.5)
            
            utility = utility_func(local_jobs, {'slope': stakeholder.preferences.get('jobs_weight', 0.3)})
            utility += utility_func(traffic_reduction, {'slope': stakeholder.preferences.get('traffic_weight', 0.4)})
            utility += utility_func(accessibility, {'slope': stakeholder.preferences.get('access_weight', 0.3)})
        
        elif stakeholder.stakeholder_type == StakeholderType.EMPLOYEE:
            # Employees value job security, fair wages, and working conditions
            job_security = solution_metrics.get('job_security', 0.5)
            fair_wages = solution_metrics.get('wage_fairness', 0.5)
            working_conditions = solution_metrics.get('working_conditions', 0.5)
            
            utility = utility_func(job_security, {'slope': stakeholder.preferences.get('security_weight', 0.4)})
            utility += utility_func(fair_wages, {'slope': stakeholder.preferences.get('wage_weight', 0.3)})
            utility += utility_func(working_conditions, {'slope': stakeholder.preferences.get('conditions_weight', 0.3)})
        
        elif stakeholder.stakeholder_type == StakeholderType.INVESTOR:
            # Investors value profitability, growth, and risk management
            profitability = solution_metrics.get('profitability', 0.5)
            growth = solution_metrics.get('growth_potential', 0.5)
            risk_management = solution_metrics.get('risk_management', 0.5)
            
            utility = utility_func(profitability, {'scale': stakeholder.preferences.get('profit_weight', 0.5)})
            utility += utility_func(growth, {'scale': stakeholder.preferences.get('growth_weight', 0.3)})
            utility += utility_func(risk_management, {'scale': stakeholder.preferences.get('risk_weight', 0.2)})
        
        elif stakeholder.stakeholder_type == StakeholderType.REGULATOR:
            # Regulators value compliance, safety, and public interest
            compliance = solution_metrics.get('regulatory_compliance', 0.5)
            safety = solution_metrics.get('safety_score', 0.5)
            public_interest = solution_metrics.get('public_interest', 0.5)
            
            utility = utility_func(compliance, {'slope': stakeholder.preferences.get('compliance_weight', 0.4)})
            utility += utility_func(safety, {'slope': stakeholder.preferences.get('safety_weight', 0.4)})
            utility += utility_func(public_interest, {'slope': stakeholder.preferences.get('public_weight', 0.2)})
        
        elif stakeholder.stakeholder_type == StakeholderType.SOCIETY:
            # Society values overall welfare, sustainability, and progress
            welfare = solution_metrics.get('social_welfare', 0.5)
            sustainability = solution_metrics.get('sustainability', 0.5)
            progress = solution_metrics.get('technological_progress', 0.5)
            
            utility = utility_func(welfare, {'scale': stakeholder.preferences.get('welfare_weight', 0.4)})
            utility += utility_func(sustainability, {'scale': stakeholder.preferences.get('sustainability_weight', 0.3)})
            utility += utility_func(progress, {'scale': stakeholder.preferences.get('progress_weight', 0.3)})
        
        return utility
    
    def calculate_total_utility(self, solution_metrics: Dict[str, float]) -> float:
        """Calculate weighted total utility across all stakeholders"""
        
        total_utility = 0.0
        stakeholder_utilities = {}
        
        for stakeholder in self.stakeholders:
            utility = self.calculate_stakeholder_utility(solution_metrics, stakeholder)
            weighted_utility = utility * stakeholder.weight
            total_utility += weighted_utility
            stakeholder_utilities[stakeholder.stakeholder_id] = utility
        
        return total_utility
    
    def generate_value_aligned_solution(self, routes: List[List[Tuple[float, float]]], 
                                        solution_metrics: Dict[str, float]) -> ValueAlignedSolution:
        """Generate a value-aligned solution with ethical compliance"""
        
        # Calculate stakeholder utilities
        stakeholder_utilities = {}
        for stakeholder in self.stakeholders:
            utility = self.calculate_stakeholder_utility(solution_metrics, stakeholder)
            stakeholder_utilities[stakeholder.stakeholder_id] = utility
        
        # Calculate total utility
        total_utility = self.calculate_total_utility(solution_metrics)
        
        # Create value-aligned solution
        solution = ValueAlignedSolution(
            routes=routes,
            stakeholder_utilities=stakeholder_utilities,
            ethical_compliance={},
            total_utility=total_utility,
            environmental_impact=solution_metrics.get('emissions_level', 50) * 10,
            social_impact=solution_metrics.get('social_impact', 50),
            economic_cost=solution_metrics.get('total_cost', 500),
            transparency_score=solution_metrics.get('transparency', 0.7),
            accountability_metrics=solution_metrics,
            ethical_reasoning=[]
        )
        
        # Evaluate ethical compliance
        solution.ethical_compliance = self.constitutional_ai.evaluate_ethical_compliance(solution)
        
        # Generate ethical reasoning
        solution.ethical_reasoning = self.constitutional_ai.generate_ethical_reasoning(solution)
        
        return solution

# Create stakeholders
stakeholders = [
    Stakeholder(
        stakeholder_id=1,
        stakeholder_type=StakeholderType.CUSTOMER,
        name="Customer Association",
        utility_function="linear",
        weight=0.25,
        preferences={"service_weight": 0.4, "time_weight": 0.3, "cost_weight": 0.3}
    ),
    Stakeholder(
        stakeholder_id=2,
        stakeholder_type=StakeholderType.ENVIRONMENT,
        name="Environmental Protection Agency",
        utility_function="concave",
        weight=0.20,
        preferences={"emissions_weight": 0.5, "renewable_weight": 0.3, "efficiency_weight": 0.2}
    ),
    Stakeholder(
        stakeholder_id=3,
        stakeholder_type=StakeholderType.COMMUNITY,
        name="Community Council",
        utility_function="linear",
        weight=0.15,
        preferences={"jobs_weight": 0.3, "traffic_weight": 0.4, "access_weight": 0.3}
    ),
    Stakeholder(
        stakeholder_id=4,
        stakeholder_type=StakeholderType.EMPLOYEE,
        name="Employee Union",
        utility_function="step",
        weight=0.15,
        preferences={"security_weight": 0.4, "wage_weight": 0.3, "conditions_weight": 0.3, "threshold": 0.6}
    ),
    Stakeholder(
        stakeholder_id=5,
        stakeholder_type=StakeholderType.INVESTOR,
        name="Investment Group",
        utility_function="convex",
        weight=0.15,
        preferences={"profit_weight": 0.5, "growth_weight": 0.3, "risk_weight": 0.2}
    ),
    Stakeholder(
        stakeholder_id=6,
        stakeholder_type=StakeholderType.REGULATOR,
        name="Transportation Authority",
        utility_function="linear",
        weight=0.05,
        preferences={"compliance_weight": 0.4, "safety_weight": 0.4, "public_weight": 0.2}
    ),
    Stakeholder(
        stakeholder_id=7,
        stakeholder_type=StakeholderType.SOCIETY,
        name="Public Interest Group",
        utility_function="concave",
        weight=0.05,
        preferences={"welfare_weight": 0.4, "sustainability_weight": 0.3, "progress_weight": 0.3}
    )
]

# Create stakeholder utility optimizer
utility_optimizer = StakeholderUtilityOptimizer(stakeholders, constitutional_ai)
print(f"Created stakeholder utility optimizer with {len(stakeholders)} stakeholders")
for stakeholder in stakeholders:
    print(f"  {stakeholder.name}: Weight {stakeholder.weight:.2f}, Type {stakeholder.stakeholder_type.value}")
```

```python
class EthicalEVRouter:
    """Ethical EV routing system with value-aligned optimization"""
    
    def __init__(self, utility_optimizer: StakeholderUtilityOptimizer):
        self.utility_optimizer = utility_optimizer
        
        # EV routing parameters
        self.depot_location = (0, 0)
        self.customer_locations = [
            (15, 10), (20, 18), (8, 22), (25, 6), (12, 15), (18, 25)
        ]
        self.charging_stations = [
            (8, 12), (22, 20)
        ]
        
        # EV specifications
        self.battery_capacity = 100.0  # kWh
        self.energy_consumption_rate = 0.8  # kWh per km
        self.max_range = self.battery_capacity / self.energy_consumption_rate
        
        # Ethical routing parameters
        self.social_importance = [0.8, 0.6, 0.9, 0.5, 0.7, 0.4]  # Customer social importance
        self.renewable_availability = [0.7, 0.9]  # Renewable energy at charging stations
        self.community_sensitivity = [0.3, 0.2, 0.8, 0.1, 0.5, 0.4]  # Community impact sensitivity
        
        # Solution history
        self.solution_history = []
        self.ethical_performance = []
    
    def calculate_distance(self, point1: Tuple[float, float], point2: Tuple[float, float]) -> float:
        """Calculate Euclidean distance between two points"""
        return np.sqrt((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2)
    
    def calculate_route_metrics(self, route: List[Tuple[float, float]]) -> Dict[str, float]:
        """Calculate comprehensive metrics for a route"""
        
        if not route:
            return {
                'total_distance': 0,
                'energy_consumed': 0,
                'emissions': 0,
                'cost': 0,
                'social_impact': 0,
                'renewable_usage': 0
            }
        
        total_distance = 0
        total_energy = 0
        total_emissions = 0
        total_cost = 0
        social_impact_score = 0
        renewable_energy_used = 0
        
        # Calculate route metrics
        for i in range(len(route) - 1):
            distance = self.calculate_distance(route[i], route[i + 1])
            total_distance += distance
            energy = distance * self.energy_consumption_rate
            total_energy += energy
            
            # Emissions (simplified model)
            emissions = energy * 0.5  # kg CO2 per kWh
            total_emissions += emissions
            
            # Cost (simplified model)
            cost = distance * 0.1 + energy * 0.15
            total_cost += cost
            
            # Social impact based on customer importance
            if i < len(self.social_importance):
                social_impact_score += self.social_importance[i] * distance
        
        # Check if route passes near charging stations
        for station_idx, station in enumerate(self.charging_stations):
            for point in route:
                if self.calculate_distance(point, station) < 5:  # Within 5 km
                    renewable_energy_used += self.renewable_availability[station_idx] * 10
        
        return {
            'total_distance': total_distance,
            'energy_consumed': total_energy,
            'emissions': total_emissions,
            'cost': total_cost,
            'social_impact': social_impact_score / max(1, len(route)),
            'renewable_usage': min(1, renewable_energy_used / max(1, total_energy))
        }
    
    def generate_ethical_routes(self) -> List[List[Tuple[float, float]]]:
        """Generate routes that consider ethical factors"""
        
        routes = []
        remaining_customers = self.customer_locations.copy()
        
        # Create routes for 2 vehicles
        for vehicle in range(2):
            if not remaining_customers:
                break
            
            route = [self.depot_location]
            current_location = self.depot_location
            battery_level = self.battery_capacity
            
            while remaining_customers and battery_level > 20:  # Keep 20% battery buffer
                # Find next customer based on ethical criteria
                best_customer = None
                best_score = -float('inf')
                best_idx = -1
                
                for idx, customer in enumerate(remaining_customers):
                    distance = self.calculate_distance(current_location, customer)
                    energy_needed = distance * self.energy_consumption_rate
                    
                    if energy_needed > battery_level - 20:
                        continue  # Not enough battery
                    
                    # Calculate ethical score
                    social_score = self.social_importance[idx] * 0.4
                    efficiency_score = (1 / (1 + distance)) * 0.3
                    community_score = (1 - self.community_sensitivity[idx]) * 0.3
                    
                    total_score = social_score + efficiency_score + community_score
                    
                    if total_score > best_score:
                        best_score = total_score
                        best_customer = customer
                        best_idx = idx
                
                if best_customer is None:
                    break
                
                # Add customer to route
                route.append(best_customer)
                remaining_customers.pop(best_idx)
                
                # Update battery level
                distance = self.calculate_distance(current_location, best_customer)
                battery_level -= distance * self.energy_consumption_rate
                current_location = best_customer
            
            # Return to depot
            route.append(self.depot_location)
            routes.append(route)
        
        return routes
    
    def optimize_ethical_routing(self) -> ValueAlignedSolution:
        """Optimize routing with ethical considerations"""
        
        print("Generating ethical EV routes...")
        routes = self.generate_ethical_routes()
        
        # Calculate comprehensive metrics
        total_metrics = {
            'total_distance': 0,
            'energy_consumed': 0,
            'emissions_level': 0,
            'total_cost': 0,
            'social_impact': 0,
            'renewable_energy': 0,
            'service_quality': 0.8,
            'timeliness': 0.85,
            'cost_factor': 0.6,
            'local_employment': 0.7,
            'traffic_reduction': 0.6,
            'accessibility': 0.8,
            'job_security': 0.9,
            'wage_fairness': 0.8,
            'working_conditions': 0.85,
            'profitability': 0.7,
            'growth_potential': 0.6,
            'risk_management': 0.8,
            'regulatory_compliance': 0.95,
            'safety_score': 0.9,
            'public_interest': 0.8,
            'social_welfare': 0.75,
            'sustainability': 0.8,
            'technological_progress': 0.7,
            'transparency': 0.85,
            'energy_efficiency': 0.8,
            'cost_efficiency': 0.75,
            'resource_utilization': 0.8,
            'explanation_quality': 0.9,
            'audit_completeness': 0.85,
            'reasoning_transparency': 0.8,
            'compliance_rate': 0.9,
            'feedback_response': 0.8,
            'violation_resolution': 0.85,
            'service_equity': 0.8,
            'community_impact': 0.7
        }
        
        # Aggregate route metrics
        for route in routes:
            route_metrics = self.calculate_route_metrics(route)
            total_metrics['total_distance'] += route_metrics['total_distance']
            total_metrics['energy_consumed'] += route_metrics['energy_consumed']
            total_metrics['emissions_level'] += route_metrics['emissions']
            total_metrics['total_cost'] += route_metrics['cost']
            total_metrics['social_impact'] += route_metrics['social_impact']
            total_metrics['renewable_energy'] += route_metrics['renewable_usage']
        
        # Normalize metrics
        total_metrics['emissions_level'] = min(1, total_metrics['emissions_level'] / 100)
        total_metrics['renewable_energy'] = total_metrics['renewable_energy'] / max(1, len(routes))
        total_metrics['social_impact'] = min(1, total_metrics['social_impact'] / 10)
        
        # Generate value-aligned solution
        solution = self.utility_optimizer.generate_value_aligned_solution(routes, total_metrics)
        
        # Calculate ethical penalty
        ethical_penalty = self.utility_optimizer.constitutional_ai.calculate_ethical_penalty(solution)
        
        # Adjust total utility based on ethical penalty
        if ethical_penalty != float('inf'):
            solution.total_utility -= ethical_penalty
        else:
            solution.total_utility = 0  # Infeasible solution
        
        # Store solution
        self.solution_history.append(solution)
        
        return solution

# Create ethical EV router
ethical_router = EthicalEVRouter(utility_optimizer)
print(f"Created ethical EV router with {len(ethical_router.customer_locations)} customers and {len(ethical_router.charging_stations)} charging stations")
```

```python
# Run the ethical EV routing optimization
print("Ethical EV Routing Optimization")
print("=" * 50)

# Optimize routes with ethical considerations
ethical_solution = ethical_router.optimize_ethical_routing()

print(f"\nOptimization Results:")
print(f"Number of routes: {len(ethical_solution.routes)}")
print(f"Total utility: {ethical_solution.total_utility:.2f}")
print(f"Environmental impact: {ethical_solution.environmental_impact:.2f} kg CO2")
print(f"Social impact score: {ethical_solution.social_impact:.2f}")
print(f"Economic cost: ${ethical_solution.economic_cost:.2f}")
print(f"Transparency score: {ethical_solution.transparency_score:.2f}")

print(f"\nStakeholder Utilities:")
for stakeholder_id, utility in ethical_solution.stakeholder_utilities.items():
    stakeholder = next(s for s in stakeholders if s.stakeholder_id == stakeholder_id)
    print(f"  {stakeholder.name}: {utility:.3f}")

print(f"\nEthical Compliance Scores:")
for principle, compliance in ethical_solution.ethical_compliance.items():
    status = "Excellent" if compliance > 0.8 else "Good" if compliance > 0.6 else "Needs Improvement"
    print(f"  {principle.value}: {compliance:.3f} ({status})")

print(f"\nEthical Reasoning:")
for i, reasoning in enumerate(ethical_solution.ethical_reasoning, 1):
    print(f"  {i}. {reasoning}")

# Calculate ethical penalty
ethical_penalty = constitutional_ai.calculate_ethical_penalty(ethical_solution)
if ethical_penalty == float('inf'):
    print(f"\nEthical Penalty: INFEASIBLE (violates hard constraints)")
else:
    print(f"\nEthical Penalty: {ethical_penalty:.3f}")

# Route details
print(f"\nRoute Details:")
for i, route in enumerate(ethical_solution.routes):
    route_metrics = ethical_router.calculate_route_metrics(route)
    print(f"\nRoute {i+1}:")
    print(f"  Stops: {len(route)}")
    print(f"  Distance: {route_metrics['total_distance']:.2f} km")
    print(f"  Energy: {route_metrics['energy_consumed']:.2f} kWh")
    print(f"  Emissions: {route_metrics['emissions']:.2f} kg CO2")
    print(f"  Cost: ${route_metrics['cost']:.2f}")
    print(f"  Renewable usage: {route_metrics['renewable_usage']:.1%}")
    print(f"  Social impact: {route_metrics['social_impact']:.2f}")
```

```python
def compare_with_utilitarian_approach():
    """Compare ethical approach with utilitarian (profit-only) approach"""
    
    print("\nComparison with Utilitarian Approach")
    print("=" * 50)
    
    # Simulate utilitarian solution (profit-only optimization)
    utilitarian_solution = {
        'total_distance': 120,
        'energy_consumed': 96,
        'emissions_level': 48,
        'total_cost': 300,
        'social_impact': 30,
        'renewable_energy': 0.2,
        'profit': 200,
        'service_quality': 0.6,
        'timeliness': 0.7,
        'cost_factor': 0.3,
        'local_employment': 0.4,
        'traffic_reduction': 0.3,
        'accessibility': 0.6,
        'job_security': 0.7,
        'wage_fairness': 0.6,
        'working_conditions': 0.7,
        'profitability': 0.9,
        'growth_potential': 0.8,
        'risk_management': 0.6,
        'regulatory_compliance': 0.7,
        'safety_score': 0.7,
        'public_interest': 0.5,
        'social_welfare': 0.4,
        'sustainability': 0.3,
        'technological_progress': 0.6,
        'transparency': 0.4,
        'energy_efficiency': 0.6,
        'cost_efficiency': 0.9,
        'resource_utilization': 0.9,
        'explanation_quality': 0.3,
        'audit_completeness': 0.4,
        'reasoning_transparency': 0.3,
        'compliance_rate': 0.7,
        'feedback_response': 0.4,
        'violation_resolution': 0.5,
        'service_equity': 0.4,
        'community_impact': 0.3
    }
    
    # Calculate utilitarian stakeholder utilities (only investor matters)
    utilitarian_utilities = {}
    for stakeholder in stakeholders:
        if stakeholder.stakeholder_type == StakeholderType.INVESTOR:
            utility = utility_optimizer.calculate_stakeholder_utility(utilitarian_solution, stakeholder)
        else:
            utility = 0.1  # Minimal utility for non-investors
        utilitarian_utilities[stakeholder.stakeholder_id] = utility
    
    utilitarian_total_utility = sum(util * stakeholder.weight 
                                  for stakeholder, util in zip(stakeholders, utilitarian_utilities.values()))
    
    # Create comparison table
    print(f"{'Metric':<25} {'Ethical Approach':<15} {'Utilitarian':<15} {'Difference':<15}")
    print("-" * 70)
    
    metrics_to_compare = [
        ('Total Utility', ethical_solution.total_utility, utilitarian_total_utility),
        ('Environmental Impact', ethical_solution.environmental_impact, utilitarian_solution['emissions_level'] * 10),
        ('Social Impact', ethical_solution.social_impact, utilitarian_solution['social_impact']),
        ('Economic Cost', ethical_solution.economic_cost, utilitarian_solution['total_cost']),
        ('Renewable Energy', ethical_solution.accountability_metrics.get('renewable_energy', 0.7), utilitarian_solution['renewable_energy']),
        ('Transparency', ethical_solution.transparency_score, utilitarian_solution['transparency']),
        ('Service Equity', ethical_solution.ethical_compliance.get(EthicalPrinciple.SOCIAL_EQUITY, 0.7), 0.4),
        ('Profit Margin', ethical_solution.accountability_metrics.get('profitability', 0.7), utilitarian_solution['profitability'])
    ]
    
    for metric, ethical_value, utilitarian_value in metrics_to_compare:
        if metric == 'Environmental Impact' or metric == 'Economic Cost':
            # Lower is better
            diff = ((utilitarian_value - ethical_value) / utilitarian_value * 100) if utilitarian_value > 0 else 0
            diff_str = f"{diff:+.1f}%"
        else:
            # Higher is better
            diff = ((ethical_value - utilitarian_value) / utilitarian_value * 100) if utilitarian_value > 0 else 0
            diff_str = f"{diff:+.1f}%"
        
        print(f"{metric:<25} {ethical_value:<15.2f} {utilitarian_value:<15.2f} {diff_str:<15}")
    
    # Stakeholder utility comparison
    print(f"\nStakeholder Utility Comparison:")
    print(f"{'Stakeholder':<25} {'Ethical':<10} {'Utilitarian':<12} {'Improvement':<12}")
    print("-" * 60)
    
    for stakeholder in stakeholders:
        ethical_util = ethical_solution.stakeholder_utilities[stakeholder.stakeholder_id]
        utilitarian_util = utilitarian_utilities[stakeholder.stakeholder_id]
        improvement = ((ethical_util - utilitarian_util) / utilitarian_util * 100) if utilitarian_util > 0 else 0
        
        print(f"{stakeholder.name:<25} {ethical_util:<10.3f} {utilitarian_util:<12.3f} {improvement:+.1f}%")
    
    # Ethical compliance comparison
    print(f"\nEthical Compliance Analysis:")
    print(f"{'Principle':<25} {'Ethical Score':<15} {'Utilitarian Score':<18} {'Improvement':<12}")
    print("-" * 70)
    
    # Calculate utilitarian ethical compliance
    utilitarian_solution_obj = ValueAlignedSolution(
        routes=[],
        stakeholder_utilities=utilitarian_utilities,
        ethical_compliance={},
        total_utility=utilitarian_total_utility,
        environmental_impact=utilitarian_solution['emissions_level'] * 10,
        social_impact=utilitarian_solution['social_impact'],
        economic_cost=utilitarian_solution['total_cost'],
        transparency_score=utilitarian_solution['transparency'],
        accountability_metrics=utilitarian_solution,
        ethical_reasoning=[]
    )
    
    utilitarian_compliance = constitutional_ai.evaluate_ethical_compliance(utilitarian_solution_obj)
    
    for principle in EthicalPrinciple:
        ethical_score = ethical_solution.ethical_compliance.get(principle, 0.5)
        utilitarian_score = utilitarian_compliance.get(principle, 0.3)
        improvement = ((ethical_score - utilitarian_score) / utilitarian_score * 100) if utilitarian_score > 0 else 0
        
        print(f"{principle.value:<25} {ethical_score:<15.3f} {utilitarian_score:<18.3f} {improvement:+.1f}%")
    
    return {
        'ethical_solution': ethical_solution,
        'utilitarian_solution': utilitarian_solution,
        'ethical_utilities': ethical_solution.stakeholder_utilities,
        'utilitarian_utilities': utilitarian_utilities
    }

# Run comparison
comparison_results = compare_with_utilitarian_approach()
```

```python
def create_ethical_audit_report(solution: ValueAlignedSolution) -> EthicalAudit:
    """Create comprehensive ethical audit report"""
    
    print("\nCreating Ethical Audit Report...")
    print("=" * 40)
    
    # Calculate compliance score
    compliance_scores = list(solution.ethical_compliance.values())
    overall_compliance = np.mean(compliance_scores)
    
    # Identify violations
    violations = []
    for principle, compliance in solution.ethical_compliance.items():
        if compliance < 0.6:  # Below acceptable threshold
            violations.append(f"{principle.value}: Compliance {compliance:.3f} below threshold")
    
    # Generate stakeholder feedback (simulated)
    stakeholder_feedback = {}
    for stakeholder in stakeholders:
        utility = solution.stakeholder_utilities[stakeholder.stakeholder_id]
        if utility > 0.8:
            feedback = "Highly satisfied with the ethical approach"
        elif utility > 0.6:
            feedback = "Satisfied with room for improvement"
        elif utility > 0.4:
            feedback = "Moderately satisfied, concerns remain"
        else:
            feedback = "Dissatisfied, requires immediate attention"
        
        stakeholder_feedback[stakeholder.stakeholder_id] = feedback
    
    # Generate recommendations
    recommendations = []
    
    if solution.ethical_compliance.get(EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY, 0) < 0.7:
        recommendations.append("Increase renewable energy usage in charging stations")
        recommendations.append("Optimize routes to reduce total distance and emissions")
    
    if solution.ethical_compliance.get(EthicalPrinciple.SOCIAL_EQUITY, 0) < 0.7:
        recommendations.append("Improve service accessibility for underserved communities")
        recommendations.append("Balance social impact across all customer segments")
    
    if solution.ethical_compliance.get(EthicalPrinciple.TRANSPARENCY, 0) < 0.8:
        recommendations.append("Enhance explanation quality for routing decisions")
        recommendations.append("Provide more detailed ethical reasoning documentation")
    
    if solution.ethical_compliance.get(EthicalPrinciple.ACCOUNTABILITY, 0) < 0.8:
        recommendations.append("Implement more robust compliance tracking")
        recommendations.append("Improve feedback response mechanisms")
    
    if not recommendations:
        recommendations.append("Maintain current ethical performance standards")
        recommendations.append("Continue monitoring for continuous improvement")
    
    # Calculate ethical score improvement (vs utilitarian baseline)
    utilitarian_baseline = 0.5  # Assumed baseline compliance
    ethical_improvement = (overall_compliance - utilitarian_baseline) / utilitarian_baseline * 100
    
    # Create audit report
    audit = EthicalAudit(
        audit_id=f"audit_{datetime.now().strftime('%Y%m%d_%H%M%S')}",
        timestamp=datetime.now(),
        solution=solution,
        compliance_score=overall_compliance,
        violations=violations,
        stakeholder_feedback=stakeholder_feedback,
        recommendations=recommendations,
        ethical_score_improvement=ethical_improvement
    )
    
    # Display audit report
    print(f"\nEthical Audit Report")
    print(f"Audit ID: {audit.audit_id}")
    print(f"Timestamp: {audit.timestamp.strftime('%Y-%m-%d %H:%M:%S')}")
    print(f"Overall Compliance Score: {audit.compliance_score:.3f}")
    print(f"Ethical Improvement: {audit.ethical_score_improvement:+.1f}% vs baseline")
    
    if audit.violations:
        print(f"\nViolations ({len(audit.violations)}):")
        for violation in audit.violations:
            print(f"  • {violation}")
    else:
        print(f"\nViolations: None - All ethical principles met minimum standards")
    
    print(f"\nStakeholder Feedback:")
    for stakeholder_id, feedback in audit.stakeholder_feedback.items():
        stakeholder = next(s for s in stakeholders if s.stakeholder_id == stakeholder_id)
        print(f"  {stakeholder.name}: {feedback}")
    
    print(f"\nRecommendations:")
    for i, recommendation in enumerate(audit.recommendations, 1):
        print(f"  {i}. {recommendation}")
    
    return audit

# Create ethical audit report
ethical_audit = create_ethical_audit_report(ethical_solution)
```
