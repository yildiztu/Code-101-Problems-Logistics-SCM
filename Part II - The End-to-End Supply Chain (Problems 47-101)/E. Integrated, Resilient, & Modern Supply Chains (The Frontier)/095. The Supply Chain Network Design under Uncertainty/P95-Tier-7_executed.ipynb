# 95. The Supply Chain Network Design under Uncertainty Problem
## Tier 7: Human-AI Symbiotic Partnership (The Centaur Model)

### Key assumptions
- Human expertise and AI computational power are complementary, not competitive
- Explainable AI (XAI) provides transparent decision rationale for human understanding
- Human-AI collaborative decision framework with confidence-based intervention
- Learning loop where AI adapts from human feedback and expertise
- Multi-expert consensus building with AI-generated alternatives

### Approach (step-by-step)
1. **Joint Problem Framing**: Human experts provide strategic context and constraints
2. **AI-Augmented Analysis**: AI generates and evaluates thousands of feasible network configurations
3. **Explainable Recommendations**: AI provides clear rationale for each network design recommendation
4. **Interactive Refinement**: Humans explore "what-if" scenarios with instant AI recalculation
5. **Consensus Building**: Multiple human experts collaborate with AI to reach optimal decisions
6. **Learning Loop**: AI learns from human interventions to improve future recommendations
7. **Trust Development**: Performance tracking and transparency build human-AI trust relationship
8. **Continuous Improvement**: System becomes more effective through ongoing collaboration

### What to look for in the results
- Human-AI collaboration metrics showing partnership effectiveness
- Explainable AI outputs with clear decision rationale
- Trust development and confidence scores over time
- Learning improvement from human feedback integration

### Concrete example (from the source)
MegaCorp Logistics Southeast Asian market entry:
- 5 human experts + 3 AI agents collaborative decision-making
- 47 potential facility locations across 8 countries analyzed
- 1,247 feasible network configurations generated and evaluated
- 23 performance metrics under 156 uncertainty scenarios
- 94% executive approval rate through transparent rationale
- 45% learning improvement from 78 human interventions

### Visualization(s)
- Human-AI collaboration workflow and decision process
- Explainable AI recommendation breakdown and confidence scores
- Trust development and learning curves over time
- Consensus building and decision quality metrics

### Why this Tier exists vs Tiers 1-5
Tier 7 addresses the critical need for human strategic judgment combined with AI computational power, creating symbiotic partnerships that leverage the unique strengths of both human and artificial intelligence.

### Pros / Cons vs Tiers 1-5
**Pros:**
- Combines human strategic expertise with AI computational capabilities
- Explainable decisions build trust and enable informed human judgment
- Learning system improves performance through human-AI interaction
- Handles complex qualitative factors that AI cannot quantify

**Cons:**
- Requires skilled human experts and significant training
- Decision process may be slower than fully automated systems
- Integration complexity between human and AI systems
- Dependence on human availability and expertise quality

### When to use this Tier
- Strategic decisions requiring human judgment and domain expertise
- Complex problems with qualitative factors beyond quantitative analysis
- High-stakes decisions where explainability and accountability are critical
- Organizations investing in human-AI collaboration capabilities

```python
try:
    @dataclass
    class HumanAICollaborationInstance:
        """Data structure for human-AI collaboration supply chain network design"""
        facilities: List[int]
        customers: List[int]
        products: List[int]
        scenarios: List[int]
        fixed_costs: List[float]
        capacities: List[float]
        transport_costs: Dict[Tuple[int, int, int], float]
        scenario_probs: List[float]
        human_experts: List[str]
        strategic_factors: Dict[str, float]
    
    def create_human_ai_instance():
        """Create a comprehensive instance for human-AI collaboration"""
        facilities = list(range(12))  # 12 potential facilities
        customers = list(range(20))   # 20 customer zones
        products = list(range(5))    # 5 product categories
        scenarios = list(range(8))    # 8 uncertainty scenarios
        
        # Fixed costs for facilities (varying by location and size)
        fixed_costs = [
            180, 220, 160, 250, 200, 170,  # Facilities 0-5
            210, 190, 240, 150, 205, 215   # Facilities 6-11
        ]
        
        # Capacities
        capacities = [
            150, 180, 130, 200, 160, 140,  # Facilities 0-5
            170, 150, 190, 120, 155, 165   # Facilities 6-11
        ]
        
        # Transportation costs (facility, customer, product)
        transport_costs = {}
        for facility in facilities:
            for customer in customers:
                for product in products:
                    # Base cost with distance and product complexity factors
                    distance_factor = abs(facility - customer // 2) * 0.8
                    product_factor = 1.0 + product * 0.15
                    base_cost = 2.5 + distance_factor + product_factor
                    transport_costs[(facility, customer, product)] = max(1.8, base_cost + np.random.normal(0, 0.3))
        
        # Scenario probabilities
        scenario_probs = [0.15, 0.20, 0.25, 0.15, 0.10, 0.05, 0.05, 0.05]
        
        # Human experts (based on source material)
        human_experts = [
            "Supply Chain Strategist",
            "Regional Operations Manager", 
            "Financial Analyst",
            "Risk Management Specialist",
            "Market Intelligence Expert"
        ]
        
        # Strategic factors that humans consider (qualitative factors)
        strategic_factors = {
            'government_stability': 0.8,
            'market_growth_potential': 0.9,
            'competitive_positioning': 0.7,
            'infrastructure_quality': 0.85,
            'labor_availability': 0.75,
            'regulatory_environment': 0.7,
            'cultural_alignment': 0.8,
            'expansion_readiness': 0.85
        }
        
        return HumanAICollaborationInstance(
            facilities=facilities,
            customers=customers,
            products=products,
            scenarios=scenarios,
            fixed_costs=fixed_costs,
            capacities=capacities,
            transport_costs=transport_costs,
            scenario_probs=scenario_probs,
            human_experts=human_experts,
            strategic_factors=strategic_factors
        )
    
    # Create the human-AI collaboration instance
    human_ai_instance = create_human_ai_instance()
    print(f"Human-AI Collaboration Instance:")
    print(f"Facilities: {len(human_ai_instance.facilities)}")
    print(f"Customers: {len(human_ai_instance.customers)}")
    print(f"Products: {len(human_ai_instance.products)}")
    print(f"Scenarios: {len(human_ai_instance.scenarios)}")
    print(f"Human experts: {len(human_ai_instance.human_experts)}")
    print(f"\nStrategic factors considered:")
    for factor, score in human_ai_instance.strategic_factors.items():
        print(f"  {factor}: {score:.2f}")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
class ExplainableAI:
    """Explainable AI system for transparent network design recommendations"""
    
    def __init__(self, instance: HumanAICollaborationInstance):
        self.instance = instance
        self.confidence_threshold = 0.7
        self.explainability_level = "detailed"  # "basic", "detailed", "expert"
        
        # Learning from human feedback
        self.human_feedback_history = []
        self.preference_weights = {
            'cost': 0.3,
            'risk': 0.2,
            'service': 0.2,
            'strategic': 0.3
        }
    
    def generate_network_alternatives(self, num_alternatives: int = 10) -> List[Dict]:
        """Generate multiple network design alternatives with explanations"""
        alternatives = []
        
        for i in range(num_alternatives):
            # Generate a network configuration
            config = self._generate_network_configuration(i)
            
            # Calculate performance metrics
            metrics = self._calculate_metrics(config)
            
            # Generate explanations
            explanation = self._generate_explanation(config, metrics, i)
            
            # Calculate confidence scores
            confidence = self._calculate_confidence(config, metrics)
            
            alternatives.append({
                'id': i,
                'configuration': config,
                'metrics': metrics,
                'explanation': explanation,
                'confidence': confidence,
                'strategic_alignment': self._calculate_strategic_alignment(config)
            })
        
        return alternatives
    
    def _generate_network_configuration(self, seed: int) -> Dict:
        """Generate a network configuration"""
        np.random.seed(seed + 100)  # Ensure reproducibility
        
        # Determine which facilities to open
        num_open = np.random.randint(3, 8)  # 3-7 facilities
        open_facilities = np.random.choice(self.instance.facilities, num_open, replace=False)
        
        # Assign customers to facilities
        customer_assignments = {}
        for customer in self.instance.customers:
            # Assign to nearest open facility with some randomness
            distances = [(abs(f - customer // 2), f) for f in open_facilities]
            distances.sort()
            
            # Choose from top 3 nearest facilities
            top_facilities = [f for _, f in distances[:3]]
            chosen_facility = np.random.choice(top_facilities)
            customer_assignments[customer] = chosen_facility
        
        return {
            'open_facilities': set(open_facilities),
            'customer_assignments': customer_assignments
        }
    
    def _calculate_metrics(self, config: Dict) -> Dict:
        """Calculate performance metrics for a configuration"""
        open_facilities = config['open_facilities']
        customer_assignments = config['customer_assignments']
        
        # Fixed costs
        fixed_cost = sum(self.instance.fixed_costs[f] for f in open_facilities)
        
        # Variable costs (simplified calculation)
        variable_cost = 0
        facility_loads = {f: 0 for f in open_facilities}
        
        for customer in self.instance.customers:
            facility = customer_assignments[customer]
            if facility in open_facilities:
                for product in self.instance.products:
                    demand = np.random.uniform(50, 150)  # Simulated demand
                    cost = self.instance.transport_costs[(facility, customer, product)]
                    variable_cost += cost * demand
                    facility_loads[facility] += demand
        
        # Expected cost across scenarios
        expected_cost = fixed_cost + variable_cost
        
        # Risk metrics
        facility_utilizations = {}
        for facility in open_facilities:
            capacity = self.instance.capacilities[facility]
            utilization = facility_loads[facility] / capacity
            facility_utilizations[facility] = utilization
        
        avg_utilization = np.mean(list(facility_utilizations.values()))
        max_utilization = max(facility_utilizations.values())
        utilization_variance = np.var(list(facility_utilizations.values()))
        
        # Service level (simplified)
        service_level = min(1.0, 1.0 - utilization_variance * 2)
        
        # Risk score (higher utilization variance = higher risk)
        risk_score = utilization_variance + (max_utilization - 0.8) * 2 if max_utilization > 0.8 else utilization_variance
        
        return {
            'fixed_cost': fixed_cost,
            'variable_cost': variable_cost,
            'expected_cost': expected_cost,
            'avg_utilization': avg_utilization,
            'max_utilization': max_utilization,
            'utilization_variance': utilization_variance,
            'service_level': service_level,
            'risk_score': risk_score,
            'num_facilities': len(open_facilities)
        }
    
    def _generate_explanation(self, config: Dict, metrics: Dict, alt_id: int) -> Dict:
        """Generate detailed explanation for the network configuration"""
        explanation = {
            'summary': f"Network configuration {alt_id} with {metrics['num_facilities']} facilities",
            'key_decisions': [],
            'performance_rationale': [],
            'trade_offs': [],
            'risk_factors': [],
            'recommendations': []
        }
        
        # Key decisions
        open_facilities = config['open_facilities']
        explanation['key_decisions'].append(
            f"Selected {len(open_facilities)} facilities: {sorted(open_facilities)}"
        )
        
        # Performance rationale
        if metrics['expected_cost'] < 5000:
            explanation['performance_rationale'].append(
                "Low expected cost due to efficient facility selection and customer assignments"
            )
        
        if metrics['service_level'] > 0.9:
            explanation['performance_rationale'].append(
                "High service level achieved through balanced facility utilization"
            )
        
        # Trade-offs
        if metrics['num_facilities'] > 5:
            explanation['trade_offs'].append(
                "Higher facility count increases fixed costs but improves service levels"
            )
        else:
            explanation['trade_offs'].append(
                "Fewer facilities reduce fixed costs but may increase transportation costs"
            )
        
        if metrics['max_utilization'] > 0.9:
            explanation['trade_offs'].append(
                "High facility utilization improves efficiency but increases operational risk"
            )
        
        # Risk factors
        if metrics['risk_score'] > 0.1:
            explanation['risk_factors'].append(
                f"High risk score ({metrics['risk_score']:.3f}) due to unbalanced utilization"
            )
        
        if metrics['utilization_variance'] > 0.05:
            explanation['risk_factors'].append(
                "High utilization variance indicates potential bottlenecks"
            )
        
        # Recommendations
        if metrics['service_level'] < 0.85:
            explanation['recommendations'].append(
                "Consider adding additional facilities or rebalancing customer assignments"
            )
        
        if metrics['max_utilization'] > 0.95:
            explanation['recommendations'].append(
                "Monitor facility capacity closely and consider expansion options"
            )
        
        return explanation
    
    def _calculate_confidence(self, config: Dict, metrics: Dict) -> float:
        """Calculate confidence score for the recommendation"""
        # Base confidence from solution quality
        quality_score = 1.0
        
        # Penalize high risk
        if metrics['risk_score'] > 0.1:
            quality_score -= 0.2
        
        # Penalize poor service level
        if metrics['service_level'] < 0.8:
            quality_score -= 0.2
        
        # Penalize extreme utilization
        if metrics['max_utilization'] > 0.95 or metrics['avg_utilization'] < 0.3:
            quality_score -= 0.1
        
        # Add some randomness to simulate uncertainty
        confidence = max(0.3, min(0.95, quality_score + np.random.normal(0, 0.1)))
        
        return confidence
    
    def _calculate_strategic_alignment(self, config: Dict) -> float:
        """Calculate alignment with strategic factors"""
        # Simplified strategic alignment calculation
        open_facilities = config['open_facilities']
        
        # Assume facilities 0-5 are in high-growth regions, 6-11 in stable regions
        high_growth_facilities = len([f for f in open_facilities if f < 6])
        stable_facilities = len(open_facilities) - high_growth_facilities
        
        # Strategic value based on balance and growth potential
        strategic_score = (high_growth_facilities * 0.6 + stable_facilities * 0.4) / len(open_facilities)
        
        return strategic_score
    
    def incorporate_human_feedback(self, alternative_id: int, human_rating: int, 
                                 human_comments: str, expert_focus: str):
        """Incorporate human feedback to improve future recommendations"""
        feedback = {
            'alternative_id': alternative_id,
            'human_rating': human_rating,  # 1-10 scale
            'human_comments': human_comments,
            'expert_focus': expert_focus,
            'timestamp': time.time()
        }
        
        self.human_feedback_history.append(feedback)
        
        # Update preference weights based on feedback
        self._update_preference_weights(expert_focus, human_rating)
    
    def _update_preference_weights(self, expert_focus: str, rating: int):
        """Update preference weights based on expert feedback"""
        # Map expert focus to preference categories
        focus_mapping = {
            'Supply Chain Strategist': 'strategic',
            'Financial Analyst': 'cost',
            'Risk Management Specialist': 'risk',
            'Regional Operations Manager': 'service',
            'Market Intelligence Expert': 'strategic'
        }
        
        category = focus_mapping.get(expert_focus, 'strategic')
        
        # Adjust weights (simplified learning algorithm)
        if rating > 7:  # Positive feedback
            self.preference_weights[category] = min(0.5, self.preference_weights[category] + 0.05)
        elif rating < 5:  # Negative feedback
            self.preference_weights[category] = max(0.1, self.preference_weights[category] - 0.05)
        
        # Renormalize weights
        total_weight = sum(self.preference_weights.values())
        for key in self.preference_weights:
            self.preference_weights[key] /= total_weight

print("Explainable AI class defined successfully")
```

```python
try:
    class HumanExpert:
        """Human expert with domain expertise and decision-making preferences"""
        
        def __init__(self, name: str, expertise: str, preference_weights: Dict[str, float]):
            self.name = name
            self.expertise = expertise
            self.preference_weights = preference_weights
            
            # Expert characteristics
            self.risk_tolerance = np.random.uniform(0.3, 0.8)
            self.strategic_focus = np.random.uniform(0.6, 0.9)
            self.cost_sensitivity = np.random.uniform(0.4, 0.8)
            
            # Decision history
            self.decision_history = []
            self.confidence_history = []
        
        def evaluate_alternative(self, alternative: Dict) -> Tuple[int, str]:
            """Evaluate an AI-generated alternative and provide rating and comments"""
            metrics = alternative['metrics']
            explanation = alternative['explanation']
            confidence = alternative['confidence']
            strategic_alignment = alternative['strategic_alignment']
            
            # Calculate expert rating based on preferences
            rating = 5  # Base rating
            
            # Cost consideration
            if metrics['expected_cost'] < 4000:
                rating += self.cost_sensitivity * 2
            elif metrics['expected_cost'] > 6000:
                rating -= self.cost_sensitivity * 2
            
            # Risk consideration
            if metrics['risk_score'] < 0.05:
                rating += (1 - self.risk_tolerance) * 2
            elif metrics['risk_score'] > 0.15:
                rating -= (1 - self.risk_tolerance) * 2
            
            # Service level consideration
            if metrics['service_level'] > 0.9:
                rating += 1.5
            elif metrics['service_level'] < 0.8:
                rating -= 1.5
            
            # Strategic alignment consideration
            rating += strategic_alignment * self.strategic_focus * 2
            
            # Confidence consideration
            rating += confidence * 0.5
            
            # Add some randomness to simulate human variability
            rating += np.random.normal(0, 0.5)
            
            # Ensure rating is within 1-10 range
            rating = max(1, min(10, rating))
            
            # Generate comments based on rating and expert characteristics
            comments = self._generate_comments(rating, metrics, explanation)
            
            # Record decision
            self.decision_history.append({
                'alternative_id': alternative['id'],
                'rating': rating,
                'comments': comments,
                'timestamp': time.time()
            })
            
            self.confidence_history.append(confidence)
            
            return int(rating), comments
        
        def _generate_comments(self, rating: int, metrics: Dict, explanation: Dict) -> str:
            """Generate expert comments based on rating and analysis"""
            comments = []
            
            if rating >= 8:
                comments.append("Excellent recommendation.")
                if metrics['service_level'] > 0.9:
                    comments.append("High service level is impressive.")
                if metrics['expected_cost'] < 4500:
                    comments.append("Cost-effective solution.")
                if self.strategic_focus > 0.7:
                    comments.append("Strong strategic alignment.")
            
            elif rating >= 6:
                comments.append("Good recommendation with some concerns.")
                if metrics['risk_score'] > 0.1:
                    comments.append("Risk level needs monitoring.")
                if metrics['max_utilization'] > 0.9:
                    comments.append("High utilization may cause issues.")
            
            elif rating >= 4:
                comments.append("Acceptable but needs improvement.")
                if metrics['service_level'] < 0.85:
                    comments.append("Service level could be better.")
                if self.cost_sensitivity > 0.6 and metrics['expected_cost'] > 5000:
                    comments.append("Cost is higher than optimal.")
            
            else:
                comments.append("Significant concerns with this approach.")
                if metrics['risk_score'] > 0.15:
                    comments.append("Risk level is too high.")
                if metrics['num_facilities'] > 7:
                    comments.append("Too many facilities increase complexity.")
                if metrics['service_level'] < 0.8:
                    comments.append("Service level is unacceptable.")
            
            # Add expertise-specific comments
            if "Financial" in self.expertise and rating < 6:
                comments.append("Financial metrics don't meet our standards.")
            elif "Risk" in self.expertise and metrics['risk_score'] > 0.1:
                comments.append("Risk mitigation strategies needed.")
            elif "Strategic" in self.expertise:
                comments.append("Long-term strategic implications considered.")
            
            return " ".join(comments)
        
        def participate_in_discussion(self, alternatives: List[Dict]) -> Dict:
            """Participate in collaborative discussion about alternatives"""
            # Evaluate top 3 alternatives
            top_alternatives = sorted(alternatives, 
                                   key=lambda x: x['metrics']['expected_cost'])[:3]
            
            discussion_points = []
            
            for alt in top_alternatives:
                rating, comments = self.evaluate_alternative(alt)
                discussion_points.append({
                    'alternative_id': alt['id'],
                    'rating': rating,
                    'comments': comments,
                    'key_concerns': self._identify_concerns(alt)
                    'suggestions': self._generate_suggestions(alt)
                })
            
            return {
                'expert': self.name,
                'expertise': self.expertise,
                'discussion_points': discussion_points,
                'overall_assessment': self._provide_overall_assessment(top_alternatives)
            }
        
        def _identify_concerns(self, alternative: Dict) -> List[str]:
            """Identify specific concerns about an alternative"""
            concerns = []
            metrics = alternative['metrics']
            
            if metrics['risk_score'] > 0.12:
                concerns.append(f"High risk score ({metrics['risk_score']:.3f})")
            
            if metrics['max_utilization'] > 0.95:
                concerns.append("Near-capacity utilization")
            
            if metrics['service_level'] < 0.85:
                concerns.append("Low service level")
            
            if metrics['num_facilities'] > 8:
                concerns.append("High operational complexity")
            
            if alternative['confidence'] < 0.6:
                concerns.append("Low recommendation confidence")
            
            return concerns
        
        def _generate_suggestions(self, alternative: Dict) -> List[str]:
            """Generate improvement suggestions for an alternative"""
            suggestions = []
            metrics = alternative['metrics']
            
            if metrics['utilization_variance'] > 0.08:
                suggestions.append("Rebalance customer assignments to reduce utilization variance")
            
            if metrics['max_utilization'] > 0.9:
                suggestions.append("Consider opening additional facility or redistributing demand")
            
            if metrics['service_level'] < 0.9:
                suggestions.append("Improve service level through better facility placement")
            
            if metrics['num_facilities'] < 4:
                suggestions.append("Evaluate if additional facilities could improve service")
            
            return suggestions
        
        def _provide_overall_assessment(self, alternatives: List[Dict]) -> str:
            """Provide overall assessment of the alternatives"""
            avg_cost = np.mean([alt['metrics']['expected_cost'] for alt in alternatives])
            avg_service = np.mean([alt['metrics']['service_level'] for alt in alternatives])
            avg_risk = np.mean([alt['metrics']['risk_score'] for alt in alternatives])
            
            assessment = f"Overall, the alternatives show average cost of ${avg_cost:.0f}, "
            assessment += f"service level of {avg_service:.3f}, and risk score of {avg_risk:.3f}. "
            
            if self.cost_sensitivity > 0.6:
                assessment += "Cost optimization should be prioritized. "
            
            if self.risk_tolerance < 0.5:
                assessment += "Risk mitigation strategies are essential. "
            
            if self.strategic_focus > 0.7:
                assessment += "Long-term strategic alignment must be considered."
            
            return assessment
    
    print("Human Expert class defined successfully")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    class HumanAICollaboration:
        """Main system for human-AI collaborative decision making"""
        
        def __init__(self, instance: HumanAICollaborationInstance):
            self.instance = instance
            
            # Initialize components
            self.explainable_ai = ExplainableAI(instance)
            
            # Initialize human experts with different preferences
            self.human_experts = []
            expert_configs = [
                ("Supply Chain Strategist", "strategic", {'strategic': 0.4, 'cost': 0.2, 'risk': 0.2, 'service': 0.2}),
                ("Regional Operations Manager", "service", {'strategic': 0.2, 'cost': 0.2, 'risk': 0.2, 'service': 0.4}),
                ("Financial Analyst", "cost", {'strategic': 0.2, 'cost': 0.5, 'risk': 0.2, 'service': 0.1}),
                ("Risk Management Specialist", "risk", {'strategic': 0.2, 'cost': 0.1, 'risk': 0.6, 'service': 0.1}),
                ("Market Intelligence Expert", "strategic", {'strategic': 0.5, 'cost': 0.2, 'risk': 0.2, 'service': 0.1})
            ]
            
            for name, expertise, preferences in expert_configs:
                expert = HumanExpert(name, expertise, preferences)
                self.human_experts.append(expert)
            
            # Collaboration history
            self.collaboration_history = []
            self.decision_outcomes = []
            
            # Performance metrics
            self.trust_scores = {expert.name: 0.5 for expert in self.human_experts}
            self.ai_confidence_scores = []
            self.human_agreement_rates = []
        
        def run_collaborative_session(self, session_id: int) -> Dict:
            """Run a complete human-AI collaborative decision session"""
            print(f"\n=== Human-AI Collaborative Session {session_id} ===")
            print(f"Participants: {len(self.human_experts)} human experts + AI system")
            
            session_start = time.time()
            
            # Phase 1: AI generates alternatives
            print("\nPhase 1: AI Generating Network Alternatives...")
            alternatives = self.explainable_ai.generate_network_alternatives(num_alternatives=12)
            
            # Phase 2: Human experts evaluate alternatives
            print("Phase 2: Human Expert Evaluation...")
            expert_evaluations = []
            
            for expert in self.human_experts:
                evaluation = expert.participate_in_discussion(alternatives)
                expert_evaluations.append(evaluation)
                
                # Incorporate feedback into AI learning
                for point in evaluation['discussion_points']:
                    self.explainable_ai.incorporate_human_feedback(
                        point['alternative_id'],
                        point['rating'],
                        point['comments'],
                        expert.expertise
                    )
            
            # Phase 3: Collaborative discussion and consensus building
            print("Phase 3: Collaborative Discussion...")
            consensus_result = self._build_consensus(alternatives, expert_evaluations)
            
            # Phase 4: Final decision and learning
            print("Phase 4: Final Decision and Learning...")
            final_decision = self._make_final_decision(consensus_result, alternatives)
            
            # Record session outcome
            session_end = time.time()
            session_duration = session_end - session_start
            
            session_record = {
                'session_id': session_id,
                'duration': session_duration,
                'alternatives_generated': len(alternatives),
                'expert_evaluations': expert_evaluations,
                'consensus_result': consensus_result,
                'final_decision': final_decision,
                'trust_scores': self.trust_scores.copy(),
                'ai_confidence': np.mean([alt['confidence'] for alt in alternatives]),
                'human_agreement': self._calculate_agreement_rate(expert_evaluations)
            }
            
            self.collaboration_history.append(session_record)
            
            # Update trust scores based on outcome
            self._update_trust_scores(session_record)
            
            print(f"\nSession {session_id} completed in {session_duration:.2f} seconds")
            print(f"Final decision: Alternative {final_decision['chosen_alternative_id']}")
            print(f"Human agreement rate: {session_record['human_agreement']:.2f}")
            print(f"AI confidence: {session_record['ai_confidence']:.2f}")
            
            return session_record
        
        def _build_consensus(self, alternatives: List[Dict], expert_evaluations: List[Dict]) -> Dict:
            """Build consensus through collaborative discussion"""
            # Calculate weighted scores for each alternative
            alternative_scores = {}
            
            for alt in alternatives:
                total_score = 0
                expert_ratings = []
                
                for evaluation in expert_evaluations:
                    for point in evaluation['discussion_points']:
                        if point['alternative_id'] == alt['id']:
                            # Weight by expert trust score
                            trust_weight = self.trust_scores[evaluation['expert']]
                            weighted_rating = point['rating'] * trust_weight
                            total_score += weighted_rating
                            expert_ratings.append(point['rating'])
                
                # Add AI confidence factor
                ai_weight = 0.3  # AI gets 30% weight in final decision
                ai_contribution = alt['confidence'] * 10 * ai_weight  # Scale confidence to 1-10
                total_score += ai_contribution
                
                # Calculate agreement level
                agreement = 1.0 - (np.std(expert_ratings) / 10) if expert_ratings else 0.5
                
                alternative_scores[alt['id']] = {
                    'total_score': total_score,
                    'avg_expert_rating': np.mean(expert_ratings) if expert_ratings else 5,
                    'agreement_level': agreement,
                    'ai_confidence': alt['confidence'],
                    'strategic_alignment': alt['strategic_alignment']
                }
            
            # Sort by total score
            sorted_alternatives = sorted(alternative_scores.items(), 
                                       key=lambda x: x[1]['total_score'], reverse=True)
            
            return {
                'ranked_alternatives': sorted_alternatives,
                'top_alternative': sorted_alternatives[0] if sorted_alternatives else None,
                'consensus_strength': self._calculate_consensus_strength(alternative_scores)
            }
        
        def _calculate_consensus_strength(self, alternative_scores: Dict) -> float:
            """Calculate the strength of consensus among alternatives"""
            if len(alternative_scores) < 2:
                return 1.0
            
            scores = [data['total_score'] for data in alternative_scores.values()]
            top_score = max(scores)
            second_score = sorted(scores)[-2] if len(scores) > 1 else min(scores)
            
            # Strong consensus if top alternative is significantly better
            strength = (top_score - second_score) / top_score if top_score > 0 else 0
            
            return min(1.0, strength)
        
        def _make_final_decision(self, consensus_result: Dict, alternatives: List[Dict]) -> Dict:
            """Make final decision based on consensus and AI recommendation"""
            if not consensus_result['top_alternative']:
                return {'chosen_alternative_id': None, 'reasoning': 'No viable alternatives'}
            
            top_alt_id, top_alt_data = consensus_result['top_alternative']
            chosen_alternative = next(alt for alt in alternatives if alt['id'] == top_alt_id)
            
            # Decision reasoning
            reasoning = []
            reasoning.append(f"Chosen alternative {top_alt_id} with highest consensus score")
            reasoning.append(f"Total score: {top_alt_data['total_score']:.2f}")
            reasoning.append(f"Expert agreement: {top_alt_data['agreement_level']:.2f}")
            reasoning.append(f"AI confidence: {top_alt_data['ai_confidence']:.2f}")
            reasoning.append(f"Strategic alignment: {top_alt_data['strategic_alignment']:.2f}")
            
            if consensus_result['consensus_strength'] > 0.3:
                reasoning.append("Strong consensus among experts")
            else:
                reasoning.append("Moderate consensus - additional review recommended")
            
            return {
                'chosen_alternative_id': top_alt_id,
                'chosen_alternative': chosen_alternative,
                'reasoning': reasoning,
                'consensus_strength': consensus_result['consensus_strength']
                'decision_confidence': top_alt_data['total_score'] / 10  # Scale to 0-1
            }
        
        def _calculate_agreement_rate(self, expert_evaluations: List[Dict]) -> float:
            """Calculate agreement rate among experts"""
            if len(expert_evaluations) < 2:
                return 1.0
            
            # Get ratings for the top alternative from each expert
            ratings = []
            
            for evaluation in expert_evaluations:
                if evaluation['discussion_points']:
                    # Get rating for the first discussed alternative
                    rating = evaluation['discussion_points'][0]['rating']
                    ratings.append(rating)
            
            if not ratings:
                return 0.5
            
            # Calculate agreement as inverse of standard deviation
            std_dev = np.std(ratings)
            agreement = max(0, 1.0 - std_dev / 5)  # Normalize by max possible std dev
            
            return agreement
        
        def _update_trust_scores(self, session_record: Dict):
            """Update trust scores based on session outcomes"""
            # Update trust based on agreement and decision quality
            agreement_rate = session_record['human_agreement']
            ai_confidence = session_record['ai_confidence']
            
            # Update AI trust (simplified)
            if agreement_rate > 0.7 and ai_confidence > 0.7:
                # Increase trust in AI recommendations
                for expert_name in self.trust_scores:
                    self.trust_scores[expert_name] = min(1.0, self.trust_scores[expert_name] + 0.05)
            elif agreement_rate < 0.5:
                # Decrease trust in AI recommendations
                for expert_name in self.trust_scores:
                    self.trust_scores[expert_name] = max(0.1, self.trust_scores[expert_name] - 0.02)
            
            # Update AI confidence scores
            self.ai_confidence_scores.append(ai_confidence)
            self.human_agreement_rates.append(agreement_rate)
        
        def analyze_collaboration_performance(self, num_sessions: int = 10) -> Dict:
            """Analyze the performance of human-AI collaboration over multiple sessions"""
            print(f"\n=== HUMAN-AI COLLABORATION PERFORMANCE ANALYSIS ===")
            
            # Run multiple collaboration sessions
            print(f"Running {num_sessions} collaboration sessions...")
            
            all_sessions = []
            for session_id in range(num_sessions):
                session = self.run_collaborative_session(session_id)
                all_sessions.append(session)
            
            # Analyze performance metrics
            avg_duration = np.mean([s['duration'] for s in all_sessions])
            avg_ai_confidence = np.mean([s['ai_confidence'] for s in all_sessions])
            avg_agreement = np.mean([s['human_agreement'] for s in all_sessions])
            
            # Trust evolution
            initial_trust = 0.5
            final_trust_scores = all_sessions[-1]['trust_scores']
            trust_improvement = np.mean([final_trust_scores[expert] - initial_trust 
                                     for expert in final_trust_scores])
            
            # Learning improvement
            learning_improvement = self._calculate_learning_improvement()
            
            print(f"\nPerformance Metrics:")
            print(f"  Average session duration: {avg_duration:.2f} seconds")
            print(f"  Average AI confidence: {avg_ai_confidence:.3f}")
            print(f"  Average human agreement: {avg_agreement:.3f}")
            print(f"  Trust improvement: {trust_improvement:.3f}")
            print(f"  Learning improvement: {learning_improvement:.3f}")
            
            return {
                'sessions': all_sessions,
                'avg_duration': avg_duration,
                'avg_ai_confidence': avg_ai_confidence,
                'avg_agreement': avg_agreement,
                'trust_improvement': trust_improvement,
                'learning_improvement': learning_improvement,
                'trust_evolution': [s['trust_scores'] for s in all_sessions],
                'confidence_evolution': [s['ai_confidence'] for s in all_sessions],
                'agreement_evolution': [s['human_agreement'] for s in all_sessions]
            }
        
        def _calculate_learning_improvement(self) -> float:
            """Calculate AI learning improvement from human feedback"""
            if len(self.explainable_ai.human_feedback_history) < 10:
                return 0.0
            
            # Analyze recent feedback trends
            recent_feedback = self.explainable_ai.human_feedback_history[-10:]
            ratings = [f['human_rating'] for f in recent_feedback]
            
            # Calculate trend (simple linear regression slope)
            if len(ratings) > 1:
                x = list(range(len(ratings)))
                slope = np.polyfit(x, ratings, 1)[0]
                return max(0, slope / 10)  # Normalize to 0-1 scale
            
            return 0.0
    
    print("Human-AI Collaboration class defined successfully")
except Exception as e:
    print(f'Error in cell: {e}')
```
