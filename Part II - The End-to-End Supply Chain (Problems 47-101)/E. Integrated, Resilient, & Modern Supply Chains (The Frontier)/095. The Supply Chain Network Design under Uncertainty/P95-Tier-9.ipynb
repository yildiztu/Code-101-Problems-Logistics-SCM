```python
@dataclass
class QuantumOptimizationInstance:
    """Data structure for quantum supply chain network design"""
    facilities: List[int]
    customers: List[int]
    products: List[int]
    scenarios: List[int]
    fixed_costs: List[float]
    capacities: List[float]
    transport_costs: Dict[Tuple[int, int, int], float]
    scenario_probs: List[float]
    quantum_params: Dict

def create_quantum_instance():
    """Create a comprehensive quantum optimization instance"""
    facilities = list(range(15))  # 15 facilities (larger than previous tiers)
    customers = list(range(30))   # 30 customers
    products = list(range(5))    # 5 products
    scenarios = list(range(10))   # 10 uncertainty scenarios
    
    # Fixed costs for facilities
    fixed_costs = [
        180, 220, 160, 250, 200, 170, 210, 190,  # Facilities 0-8
        240, 150, 205, 215, 195, 225, 185        # Facilities 9-14
    ]
    
    # Capacities
    capacities = [
        150, 180, 130, 200, 160, 140, 170, 150,  # Facilities 0-8
        190, 120, 155, 165, 145, 175, 135        # Facilities 9-14
    ]
    
    # Transportation costs (facility, customer, product)
    transport_costs = {}
    for facility in facilities:
        for customer in customers:
            for product in products:
                # Base cost with distance and product complexity factors
                distance_factor = abs(facility - customer // 2) * 0.8
                product_factor = 1.0 + product * 0.15
                base_cost = 2.5 + distance_factor + product_factor
                transport_costs[(facility, customer, product)] = max(1.8, base_cost + np.random.normal(0, 0.3))
    
    # Scenario probabilities
    scenario_probs = [0.12, 0.15, 0.18, 0.15, 0.10, 0.08, 0.07, 0.05, 0.03, 0.07]
    
    # Quantum parameters
    quantum_params = {
        'num_qubits': len(facilities) * len(customers),  # 450 qubits
        'circuit_depth': 50,  # 50 quantum layers
        'annealing_time': 100,  # microseconds
        'temperature_start': 1000.0,  # Initial temperature
        'temperature_end': 0.001,    # Final temperature
        'quantum_gates': ['H', 'CNOT', 'RZ', 'RX'],  # Quantum gate set
        'mixing_strength': 0.5,  # QAOA mixing parameter
        'cost_strength': 1.0    # QAOA cost parameter
    }
    
    return QuantumOptimizationInstance(
        facilities=facilities,
        customers=customers,
        products=products,
        scenarios=scenarios,
        fixed_costs=fixed_costs,
        capacities=capacities,
        transport_costs=transport_costs,
        scenario_probs=scenario_probs,
        quantum_params=quantum_params
    )

# Create the quantum optimization instance
quantum_instance = create_quantum_instance()
print(f"Quantum Optimization Instance:")
print(f"Facilities: {len(quantum_instance.facilities)}")
print(f"Customers: {len(quantum_instance.customers)}")
print(f"Products: {len(quantum_instance.products)}")
print(f"Scenarios: {len(quantum_instance.scenarios)}")
print(f"Qubits required: {quantum_instance.quantum_params['num_qubits']}")
print(f"Circuit depth: {quantum_instance.quantum_params['circuit_depth']}")
print(f"Binary variables: {len(quantum_instance.facilities) * len(quantum_instance.customers)}")
```

```python
class QuboFormulation:
    """QUBO formulation for supply chain network design"""
    
    def __init__(self, instance: QuantumOptimizationInstance):
        self.instance = instance
        self.num_variables = len(instance.facilities) * len(instance.customers)
        self.Q_matrix = np.zeros((self.num_variables, self.num_variables))
        self.linear_terms = np.zeros(self.num_variables)
        
        # Variable mapping: (facility, customer) -> variable index
        self.var_mapping = {}
        idx = 0
        for facility in instance.facilities:
            for customer in instance.customers:
                self.var_mapping[(facility, customer)] = idx
                idx += 1
    
    def formulate_qubo(self, penalty_weights: Dict[str, float]) -> Tuple[np.ndarray, np.ndarray]:
        """Formulate QUBO with penalty method for constraints"""
        
        # Initialize QUBO components
        self.Q_matrix = np.zeros((self.num_variables, self.num_variables))
        self.linear_terms = np.zeros(self.num_variables)
        
        # 1. Objective function: minimize total cost
        self._add_objective_terms()
        
        # 2. Constraints with penalty terms
        self._add_facility_capacity_constraints(penalty_weights['capacity'])
        self._add_customer_demand_constraints(penalty_weights['demand'])
        self._add_facility_opening_constraints(penalty_weights['opening'])
        
        return self.Q_matrix, self.linear_terms
    
    def _add_objective_terms(self):
        """Add objective function terms to QUBO"""
        
        for facility in self.instance.facilities:
            for customer in self.instance.customers:
                var_idx = self.var_mapping[(facility, customer)]
                
                # Fixed cost contribution (if facility serves any customer)
                # This is simplified - in reality would need additional binary variables
                fixed_cost_per_customer = self.instance.fixed_costs[facility] / len(self.instance.customers)
                self.linear_terms[var_idx] += fixed_cost_per_customer
                
                # Transportation cost contribution
                for product in self.instance.products:
                    transport_cost = self.instance.transport_costs.get((facility, customer, product), 5.0)
                    # Simplified demand assumption
                    demand = np.random.uniform(50, 150)
                    self.linear_terms[var_idx] += transport_cost * demand / len(self.instance.products)
    
    def _add_facility_capacity_constraints(self, penalty_weight: float):
        """Add facility capacity constraints"""
        
        for facility in self.instance.facilities:
            facility_capacity = self.instance.capacities[facility]
            
            # Get all variables for this facility
            facility_vars = []
            for customer in self.instance.customers:
                var_idx = self.var_mapping[(facility, customer)]
                facility_vars.append(var_idx)
            
            # Add penalty for exceeding capacity
            # This is simplified - would need more sophisticated constraint handling
            for i, var_i in enumerate(facility_vars):
                for j, var_j in enumerate(facility_vars):
                    if i != j:
                        # Penalty for assigning too many customers to same facility
                        self.Q_matrix[var_i, var_j] += penalty_weight * 0.1
    
    def _add_customer_demand_constraints(self, penalty_weight: float):
        """Add customer demand constraints (each customer must be served)"""
        
        for customer in self.instance.customers:
            # Get all variables for this customer
            customer_vars = []
            for facility in self.instance.facilities:
                var_idx = self.var_mapping[(facility, customer)]
                customer_vars.append(var_idx)
            
            # Add penalty for not serving customer
            for i, var_i in enumerate(customer_vars):
                # Linear penalty for not selecting any facility
                self.linear_terms[var_i] -= penalty_weight
                
                for j, var_j in enumerate(customer_vars):
                    if i != j:
                        # Penalty for serving customer from multiple facilities
                        self.Q_matrix[var_i, var_j] += penalty_weight * 2.0
    
    def _add_facility_opening_constraints(self, penalty_weight: float):
        """Add facility opening/closing constraints"""
        
        # This is simplified - would need additional binary variables for facility opening
        for facility in self.instance.facilities:
            facility_vars = []
            for customer in self.instance.customers:
                var_idx = self.var_mapping[(facility, customer)]
                facility_vars.append(var_idx)
            
            # Add penalty for opening facility with too few customers
            for var in facility_vars:
                self.linear_terms[var] += penalty_weight * 0.5
    
    def decode_solution(self, binary_solution: np.ndarray) -> Dict:
        """Decode binary solution to network configuration"""
        
        open_facilities = set()
        customer_assignments = {}
        
        for facility in self.instance.facilities:
            for customer in self.instance.customers:
                var_idx = self.var_mapping[(facility, customer)]
                if binary_solution[var_idx] == 1:
                    open_facilities.add(facility)
                    customer_assignments[customer] = facility
        
        return {
            'open_facilities': list(open_facilities),
            'customer_assignments': customer_assignments
        }
    
    def calculate_qubo_energy(self, binary_solution: np.ndarray) -> float:
        """Calculate QUBO energy for binary solution"""
        
        # Linear terms
        linear_energy = np.dot(self.linear_terms, binary_solution)
        
        # Quadratic terms
        quadratic_energy = 0
        for i in range(self.num_variables):
            for j in range(self.num_variables):
                if i != j:
                    quadratic_energy += self.Q_matrix[i, j] * binary_solution[i] * binary_solution[j]
        
        return linear_energy + quadratic_energy

print("QUBO Formulation class defined successfully")
```

```python
class QuantumSimulator:
    """Classical simulator for quantum optimization algorithms"""
    
    def __init__(self, num_qubits: int, params: Dict):
        self.num_qubits = num_qubits
        self.params = params
        self.quantum_state = None
        self.measurement_history = []
        
    def initialize_quantum_state(self) -> np.ndarray:
        """Initialize quantum state in equal superposition"""
        # Start with equal superposition of all states
        num_states = 2 ** self.num_qubits
        self.quantum_state = np.ones(num_states) / np.sqrt(num_states)
        return self.quantum_state
    
    def apply_hadamard_gate(self, qubit: int):
        """Apply Hadamard gate to specified qubit"""
        if self.quantum_state is None:
            self.initialize_quantum_state()
        
        # Simplified Hadamard gate application
        # In reality, this would be more complex quantum state manipulation
        noise = np.random.normal(0, 0.01, size=self.quantum_state.shape)
        self.quantum_state = self.quantum_state + noise
        self.quantum_state = self.quantum_state / np.linalg.norm(self.quantum_state)
    
    def apply_cnot_gate(self, control_qubit: int, target_qubit: int):
        """Apply CNOT gate (entanglement)"""
        if self.quantum_state is None:
            self.initialize_quantum_state()
        
        # Simplified CNOT gate application
        # Create entanglement between qubits
        for i in range(len(self.quantum_state)):
            if (i >> control_qubit) & 1:  # If control qubit is 1
                # Flip target qubit (simplified)
                target_bit = (i >> target_qubit) & 1
                if target_bit == 0:
                    new_i = i | (1 << target_qubit)
                else:
                    new_i = i & ~(1 << target_qubit)
                
                # Swap amplitudes (simplified entanglement)
                if new_i < len(self.quantum_state):
                    self.quantum_state[i], self.quantum_state[new_i] = \
                        self.quantum_state[new_i], self.quantum_state[i]
    
    def apply_rotation_gate(self, qubit: int, angle: float, gate_type: str = 'RZ'):
        """Apply rotation gate (RZ or RX)"""
        if self.quantum_state is None:
            self.initialize_quantum_state()
        
        # Simplified rotation gate application
        for i in range(len(self.quantum_state)):
            bit_value = (i >> qubit) & 1
            if bit_value == 1:
                # Apply phase rotation
                phase = np.exp(1j * angle)
                self.quantum_state[i] *= phase
    
    def build_quantum_circuit(self, depth: int) -> np.ndarray:
        """Build quantum circuit with specified depth"""
        print(f"Building quantum circuit with {depth} layers...")
        
        self.initialize_quantum_state()
        
        for layer in range(depth):
            # Layer 1: Hadamard gates (create superposition)
            if layer == 0:
                for qubit in range(min(self.num_qubits, 10)):  # Limit for efficiency
                    self.apply_hadamard_gate(qubit)
            
            # Layer 2: Entanglement (CNOT gates)
            if layer % 2 == 1:
                # Create entanglement between neighboring qubits
                for qubit in range(min(self.num_qubits - 1, 9)):
                    self.apply_cnot_gate(qubit, qubit + 1)
            
            # Layer 3: Rotation gates (problem-specific encoding)
            if layer % 3 == 2:
                for qubit in range(min(self.num_qubits, 10)):
                    angle = np.random.uniform(0, 2*np.pi)
                    self.apply_rotation_gate(qubit, angle, 'RZ')
            
            # Progress indicator
            if (layer + 1) % 10 == 0:
                print(f"  Circuit layer {layer + 1}/{depth} completed")
        
        return self.quantum_state
    
    def measure_quantum_state(self, num_shots: int = 1000) -> List[np.ndarray]:
        """Measure quantum state multiple times"""
        if self.quantum_state is None:
            self.initialize_quantum_state()
        
        measurements = []
        probabilities = np.abs(self.quantum_state) ** 2
        
        for shot in range(num_shots):
            # Sample from quantum state probability distribution
            measured_state = np.random.choice(range(len(probabilities)), 
                                            p=probabilities)
            
            # Convert to binary representation
            binary_solution = np.array([(measured_state >> i) & 1 
                                       for i in range(self.num_qubits)])
            measurements.append(binary_solution)
        
        self.measurement_history.extend(measurements)
        return measurements
    
    def get_measurement_statistics(self) -> Dict:
        """Get statistics from quantum measurements"""
        if not self.measurement_history:
            return {'error': 'No measurements available'}
        
        measurements_array = np.array(self.measurement_history)
        
        # Calculate frequency of each solution
        unique_solutions, counts = np.unique(measurements_array, axis=0, return_counts=True)
        
        # Get top solutions
        top_indices = np.argsort(counts)[-5:][::-1]
        top_solutions = unique_solutions[top_indices]
        top_counts = counts[top_indices]
        
        return {
            'total_measurements': len(self.measurement_history),
            'unique_solutions': len(unique_solutions),
            'top_solutions': top_solutions,
            'top_counts': top_counts,
            'measurement_probabilities': top_counts / len(self.measurement_history)
        }

print("Quantum Simulator class defined successfully")
```

```python
class QuantumAnnealingOptimizer:
    """Quantum annealing optimization algorithm"""
    
    def __init__(self, qubo_formulation: QuboFormulation, quantum_simulator: QuantumSimulator):
        self.qubo = qubo_formulation
        self.simulator = quantum_simulator
        self.optimization_history = []
        
    def quantum_annealing(self, Q_matrix: np.ndarray, linear_terms: np.ndarray,
                          temperature_schedule: List[float], num_shots: int = 1000) -> Dict:
        """Perform quantum annealing optimization"""
        
        print(f"Starting Quantum Annealing Optimization...")
        print(f"Problem size: {self.qubo.num_variables} variables")
        print(f"Temperature schedule: {len(temperature_schedule)} steps")
        print(f"Measurements per step: {num_shots}")
        
        best_solution = None
        best_energy = float('inf')
        energy_history = []
        
        for step, temperature in enumerate(temperature_schedule):
            # 1. Build quantum circuit for current temperature
            circuit_depth = int(self.simulator.params['circuit_depth'] * (1 - step / len(temperature_schedule)))
            quantum_state = self.simulator.build_quantum_circuit(circuit_depth)
            
            # 2. Apply temperature-dependent operations
            # Higher temperature = more exploration
            for qubit in range(min(self.qubo.num_variables, 10)):
                exploration_angle = temperature * 0.01
                self.simulator.apply_rotation_gate(qubit, exploration_angle, 'RX')
            
            # 3. Measure quantum state
            measurements = self.simulator.measure_quantum_state(num_shots)
            
            # 4. Evaluate measurements
            step_best_solution = None
            step_best_energy = float('inf')
            
            for measurement in measurements:
                energy = self.qubo.calculate_qubo_energy(measurement)
                
                if energy < step_best_energy:
                    step_best_energy = energy
                    step_best_solution = measurement.copy()
                
                if energy < best_energy:
                    best_energy = energy
                    best_solution = measurement.copy()
            
            energy_history.append(step_best_energy)
            
            # 5. Update quantum state based on best solution (quantum tunneling effect)
            if step_best_solution is not None:
                # Bias quantum state toward good solutions
                self._bias_quantum_state(step_best_solution, temperature)
            
            # Progress indicator
            if (step + 1) % 10 == 0:
                print(f"  Annealing step {step + 1}/{len(temperature_schedule)} - "
                      f"Best energy: {best_energy:.2f}")
        
        # Decode final solution
        final_config = self.qubo.decode_solution(best_solution)
        
        optimization_result = {
            'best_solution': best_solution,
            'best_energy': best_energy,
            'final_config': final_config,
            'energy_history': energy_history,
            'total_measurements': len(temperature_schedule) * num_shots
        }
        
        self.optimization_history.append(optimization_result)
        
        print(f"\nQuantum Annealing Completed!")
        print(f"Best energy: {best_energy:.2f}")
        print(f"Open facilities: {len(final_config['open_facilities'])}")
        print(f"Customers served: {len(final_config['customer_assignments'])}")
        
        return optimization_result
    
    def _bias_quantum_state(self, good_solution: np.ndarray, temperature: float):
        """Bias quantum state toward good solutions (quantum tunneling)"""
        # This is a simplified implementation of quantum tunneling effect
        # In reality, this would involve more sophisticated quantum operations
        
        if self.simulator.quantum_state is None:
            return
        
        # Create bias based on good solution
        for i in range(len(good_solution)):
            if good_solution[i] == 1:
                # Increase amplitude of states with bit i = 1
                for state_idx in range(len(self.simulator.quantum_state)):
                    if (state_idx >> i) & 1:
                        bias_factor = 1.0 + (1.0 / temperature) * 0.01
                        self.simulator.quantum_state[state_idx] *= bias_factor
        
        # Renormalize quantum state
        self.simulator.quantum_state = self.simulator.quantum_state / np.linalg.norm(self.simulator.quantum_state)
    
    def create_temperature_schedule(self, start_temp: float, end_temp: float, steps: int) -> List[float]:
        """Create exponential cooling schedule"""
        schedule = []
        for step in range(steps):
            # Exponential cooling
            t = start_temp * (end_temp / start_temp) ** (step / (steps - 1))
            schedule.append(t)
        return schedule

print("Quantum Annealing Optimizer class defined successfully")
```

```python
class HybridQuantumClassicalOptimizer:
    """Hybrid quantum-classical optimization algorithm"""
    
    def __init__(self, qubo_formulation: QuboFormulation, quantum_simulator: QuantumSimulator):
        self.qubo = qubo_formulation
        self.simulator = quantum_simulator
        self.quantum_optimizer = QuantumAnnealingOptimizer(qubo_formulation, quantum_simulator)
        self.optimization_history = []
    
    def hybrid_optimization(self, max_iterations: int = 10, 
                           classical_iterations: int = 100,
                           quantum_shots: int = 500) -> Dict:
        """Perform hybrid quantum-classical optimization"""
        
        print(f"Starting Hybrid Quantum-Classical Optimization...")
        print(f"Max iterations: {max_iterations}")
        print(f" Classical iterations: {classical_iterations}")
        print(f" Quantum shots: {quantum_shots}")
        
        # Initialize with classical solution
        current_solution = self._classical_local_search(classical_iterations)
        current_energy = self.qubo.calculate_qubo_energy(current_solution)
        
        best_solution = current_solution.copy()
        best_energy = current_energy
        
        optimization_log = []
        
        for iteration in range(max_iterations):
            print(f"\nIteration {iteration + 1}/{max_iterations}")
            
            # 1. Quantum phase: Use quantum annealing to escape local minima
            print(f"  Quantum phase...")
            temperature_schedule = self.quantum_optimizer.create_temperature_schedule(
                100.0, 0.1, 20
            )
            
            quantum_result = self.quantum_optimizer.quantum_annealing(
                self.qubo.Q_matrix, self.qubo.linear_terms,
                temperature_schedule, quantum_shots
            )
            
            quantum_solution = quantum_result['best_solution']
            quantum_energy = quantum_result['best_energy']
            
            # 2. Classical phase: Refine quantum solution with local search
            print(f"  Classical refinement...")
            refined_solution = self._classical_local_search(classical_iterations, quantum_solution)
            refined_energy = self.qubo.calculate_qubo_energy(refined_solution)
            
            # 3. Update best solution
            if refined_energy < best_energy:
                best_solution = refined_solution.copy()
                best_energy = refined_energy
                print(f"  New best solution found! Energy: {best_energy:.2f}")
            
            # 4. Adaptive strategy: Choose best between quantum and classical
            if quantum_energy < current_energy:
                current_solution = quantum_solution.copy()
                current_energy = quantum_energy
                improvement_type = "quantum"
            else:
                current_solution = refined_solution.copy()
                current_energy = refined_energy
                improvement_type = "classical"
            
            # Log iteration results
            iteration_log = {
                'iteration': iteration + 1,
                'quantum_energy': quantum_energy,
                'classical_energy': refined_energy,
                'best_energy': best_energy,
                'improvement_type': improvement_type
            }
            optimization_log.append(iteration_log)
            
            print(f"  Quantum energy: {quantum_energy:.2f}")
            print(f"  Classical energy: {refined_energy:.2f}")
            print(f"  Best energy: {best_energy:.2f}")
            print(f"  Improvement type: {improvement_type}")
        
        # Decode final solution
        final_config = self.qubo.decode_solution(best_solution)
        
        result = {
            'best_solution': best_solution,
            'best_energy': best_energy,
            'final_config': final_config,
            'optimization_log': optimization_log,
            'total_iterations': max_iterations
        }
        
        self.optimization_history.append(result)
        
        print(f"\nHybrid Optimization Completed!")
        print(f"Final best energy: {best_energy:.2f}")
        print(f"Open facilities: {len(final_config['open_facilities'])}")
        print(f"Total customers served: {len(final_config['customer_assignments'])}")
        
        return result
    
    def _classical_local_search(self, max_iterations: int, initial_solution: Optional[np.ndarray] = None) -> np.ndarray:
        """Classical local search for refinement"""
        
        if initial_solution is None:
            # Initialize random solution
            current_solution = np.random.randint(0, 2, self.qubo.num_variables)
        else:
            current_solution = initial_solution.copy()
        
        current_energy = self.qubo.calculate_qubo_energy(current_solution)
        
        for iteration in range(max_iterations):
            # Generate neighbor solution (flip one bit)
            neighbor_solution = current_solution.copy()
            flip_idx = np.random.randint(0, self.qubo.num_variables)
            neighbor_solution[flip_idx] = 1 - neighbor_solution[flip_idx]
            
            neighbor_energy = self.qubo.calculate_qubo_energy(neighbor_solution)
            
            # Accept if better (hill climbing)
            if neighbor_energy < current_energy:
                current_solution = neighbor_solution
                current_energy = neighbor_energy
        
        return current_solution

print("Hybrid Quantum-Classical Optimizer class defined successfully")
```

```python
def compare_with_classical_methods():
    """Compare quantum optimization with classical methods"""
    
    print(f"\n=== QUANTUM VS CLASSICAL COMPARISON ===")
    
    # 1. Classical greedy solution
    print(f"\n1. Classical Greedy Method:")
    greedy_solution = np.zeros(qubo_formulation.num_variables)
    
    # Assign each customer to nearest facility
    for customer in quantum_instance.customers:
        best_facility = None
        best_cost = float('inf')
        
        for facility in quantum_instance.facilities:
            var_idx = qubo_formulation.var_mapping[(facility, customer)]
            # Simplified cost calculation
            cost = quantum_instance.fixed_costs[facility] / len(quantum_instance.customers)
            
            if cost < best_cost:
                best_cost = cost
                best_facility = facility
        
        if best_facility is not None:
            var_idx = qubo_formulation.var_mapping[(best_facility, customer)]
            greedy_solution[var_idx] = 1
    
    greedy_energy = qubo_formulation.calculate_qubo_energy(greedy_solution)
    greedy_config = qubo_formulation.decode_solution(greedy_solution)
    
    print(f"  Energy: {greedy_energy:.2f}")
    print(f"  Open facilities: {len(greedy_config['open_facilities'])}")
    print(f"  Customers served: {len(greedy_config['customer_assignments'])}")
    
    # 2. Classical random search
    print(f"\n2. Classical Random Search:")
    best_random_solution = None
    best_random_energy = float('inf')
    
    for _ in range(1000):
        random_solution = np.random.randint(0, 2, qubo_formulation.num_variables)
        random_energy = qubo_formulation.calculate_qubo_energy(random_solution)
        
        if random_energy < best_random_energy:
            best_random_energy = random_energy
            best_random_solution = random_solution.copy()
    
    random_config = qubo_formulation.decode_solution(best_random_solution)
    
    print(f"  Best energy: {best_random_energy:.2f}")
    print(f"  Open facilities: {len(random_config['open_facilities'])}")
    print(f"  Customers served: {len(random_config['customer_assignments'])}")
    
    # 3. Quantum-hybrid solution (from previous optimization)
    print(f"\n3. Quantum-Hybrid Method:")
    print(f"  Energy: {best_energy:.2f}")
    print(f"  Open facilities: {len(final_config['open_facilities'])}")
    print(f"  Customers served: {len(final_config['customer_assignments'])}")
    
    # 4. Performance comparison
    print(f"\n4. Performance Comparison:")
    
    improvements = {
        'greedy': ((greedy_energy - best_energy) / greedy_energy) * 100,
        'random': ((best_random_energy - best_energy) / best_random_energy) * 100,
    }
    
    print(f"  Quantum vs Greedy improvement: {improvements['greedy']:.1f}%")
    print(f"  Quantum vs Random Search improvement: {improvements['random']:.1f}%")
    
    # 5. Constraint satisfaction comparison
    print(f"\n5. Constraint Satisfaction:")
    
    greedy_violations = sum(1 for customer in quantum_instance.customers 
                          if customer not in greedy_config['customer_assignments'])
    random_violations = sum(1 for customer in quantum_instance.customers 
                          if customer not in random_config['customer_assignments'])
    quantum_violations = sum(1 for customer in quantum_instance.customers 
                           if customer not in final_config['customer_assignments'])
    
    print(f"  Greedy method: {(len(quantum_instance.customers) - greedy_violations) / len(quantum_instance.customers) * 100:.1f}%")
    print(f"  Random search: {(len(quantum_instance.customers) - random_violations) / len(quantum_instance.customers) * 100:.1f}%")
    print(f"  Quantum hybrid: {(len(quantum_instance.customers) - quantum_violations) / len(quantum_instance.customers) * 100:.1f}%")
    
    return {
        'greedy_energy': greedy_energy,
        'random_energy': best_random_energy,
        'quantum_energy': best_energy,
        'improvements': improvements
    }

# Run comparison
comparison_results = compare_with_classical_methods()
```
