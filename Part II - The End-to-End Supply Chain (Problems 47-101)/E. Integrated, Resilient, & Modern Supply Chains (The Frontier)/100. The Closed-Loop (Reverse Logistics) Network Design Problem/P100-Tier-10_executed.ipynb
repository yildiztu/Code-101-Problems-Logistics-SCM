# 100. The Closed-Loop Network Design Problem
## Tier 10: The Proactive Demand & Market Shaping

### Key Assumptions
- Network design can proactively influence and shape market demand
- Customer behavior can be guided through strategic interventions
- Market dynamics are responsive to network configuration decisions
- Competitive advantage can be created through demand orchestration
- Long-term market share can be increased through strategic positioning
- Network design becomes a strategic market participation tool

### Approach (Step-by-Step)
1. **Market Analysis**: Understand current demand patterns and customer segments
2. **Demand Modeling**: Create models of customer behavior and elasticity
3. **Strategic Positioning**: Identify opportunities for market influence
4. **Network Configuration**: Design network to shape desired demand patterns
5. **Market Orchestration**: Implement strategies to guide customer behavior
6. **Performance Monitoring**: Track market share and competitive position

### What to Look for in Results
- Market share growth and competitive positioning
- Demand elasticity and customer response to network changes
- Strategic interventions and their effectiveness
- Long-term market dynamics and sustainability
- Network configuration impact on market behavior
- Return on investment from market shaping activities

We'll implement:
- Market demand modeling with customer segmentation

- Competitive positioning and market share optimization

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass, field
from typing import List, Tuple, Optional, Dict, Any
import random
from collections import defaultdict, deque
import warnings
from enum import Enum
import time
warnings.filterwarnings('ignore')

# Set style for better visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)
```

```python
class CustomerSegment(Enum):
    """Customer segments for market analysis"""
    PRICE_SENSITIVE = "price_sensitive"
    QUALITY_FOCUSED = "quality_focused"
    CONVENIENCE_SEEKING = "convenience_seeking"
    SUSTAINABILITY_CONSCIOUS = "sustainability_conscious"
    PREMIUM_SEGMENT = "premium_segment"

@dataclass
class MarketSegment:
    """Represents a market segment"""
    segment_id: str
    segment_type: CustomerSegment
    size: int  # Number of customers
    price_elasticity: float  # Price sensitivity
    quality_sensitivity: float  # Quality importance
    convenience_sensitivity: float  # Convenience importance
    sustainability_sensitivity: float  # Sustainability importance
    current_market_share: float = 0.0
    growth_potential: float = 0.0
    
    def calculate_demand_response(self, price_change: float, 
                                 quality_change: float,
                                 convenience_change: float,
                                 sustainability_change: float) -> float:
        """Calculate demand response to network changes"""
        
        # Elasticity-based demand response
        price_response = -self.price_elasticity * price_change
        quality_response = self.quality_sensitivity * quality_change
        convenience_response = self.convenience_sensitivity * convenience_change
        sustainability_response = self.sustainability_sensitivity * sustainability_change
        
        total_response = price_response + quality_response + convenience_response + sustainability_response
        
        # Apply growth potential multiplier
        response_multiplier = 1.0 + self.growth_potential * 0.5
        
        return total_response * response_multiplier

@dataclass
class StrategicIntervention:
    """Represents a strategic market intervention"""
    intervention_id: str
    intervention_type: str  # 'pricing', 'service', 'marketing', 'partnership', 'technology'
    target_segments: List[str]
    investment_cost: float
    expected_impact: Dict[str, float]  # Impact on different attributes
    implementation_time: int  # months
    success_probability: float = 0.7
    
    def calculate_roi(self, market_response: Dict[str, float]) -> float:
        """Calculate return on investment"""
        
        # Simplified ROI calculation
        total_impact = sum(market_response.values())
        roi = (total_impact * 1000 - self.investment_cost) / self.investment_cost
        
        return roi

@dataclass
class MarketShaper:
    """Market shaping and demand orchestration system"""
    shaper_id: str
    market_segments: List[MarketSegment] = field(default_factory=list)
    interventions: List[StrategicIntervention] = field(default_factory=list)
    network_configuration: Dict[str, Any] = field(default_factory=dict)
    market_history: List[Dict[str, Any]] = field(default_factory=list)
    
    def add_market_segment(self, segment: MarketSegment):
        """Add a market segment to analysis"""
        self.market_segments.append(segment)
    
    def add_intervention(self, intervention: StrategicIntervention):
        """Add a strategic intervention"""
        self.interventions.append(intervention)
    
    def configure_network(self, configuration: Dict[str, Any]):
        """Configure network for market shaping"""
        self.network_configuration = configuration
    
    def analyze_market_opportunities(self) -> Dict[str, Any]:
        """Analyze market opportunities for strategic positioning"""
        
        opportunities = {
            'high_growth_segments': [],
            'underserved_segments': [],
            'competitive_advantages': [],
            'strategic_priorities': []
            'market_size': sum(segment.size for segment in self.market_segments),
            'growth_potential': 0.0
        }
        
        # Analyze each segment
        for segment in self.market_segments:
            # High growth segments
            if segment.growth_potential > 0.3:
                opportunities['high_growth_segments'].append({
                    'segment_id': segment.segment_id,
                    'growth_potential': segment.growth_potential,
                    'size': segment.size
                })
            
            # Underserved segments (low current market share)
            if segment.current_market_share < 0.2:
                opportunities['underserved_segments'].append({
                    'segment_id': segment.segment_id,
                    'current_share': segment.current_market_share,
                    'size': segment.size
                })
            
            # Calculate total growth potential
            opportunities['growth_potential'] += segment.size * segment.growth_potential
        
        # Identify competitive advantages
        if any(segment.quality_sensitivity > 0.7 for segment in self.market_segments):
            opportunities['competitive_advantages'].append('quality_leadership')
        
        if any(segment.sustainability_sensitivity > 0.7 for segment in self.market_segments):
            opportunities['competitive_advantages'].append('sustainability_focus')
        
        if any(segment.convenience_sensitivity > 0.7 for segment in self.market_segments):
            opportunities['competitive_advantages'].append('convenience_excellence')
        
        # Generate strategic priorities
        if opportunities['high_growth_segments']:
            opportunities['strategic_priorities'].append('capture_growth_segments')
        
        if opportunities['underserved_segments']:
            opportunities['strategic_priorities'].append('expand_underserved_markets')
        
        if opportunities['competitive_advantages']:
            opportunities['strategic_priorities'].append('leverage_competitive_advantages')
        
        return opportunities
    
    def simulate_market_response(self, interventions: List[StrategicIntervention], 
                                time_horizon: int = 12) -> Dict[str, Any]:
        """Simulate market response to strategic interventions"""
        
        simulation_results = {
            'time_horizon': time_horizon,
            'monthly_responses': [],
            'final_market_share': {},
            'total_roi': 0.0,
            'successful_interventions': [],
            'market_growth': 0.0
        }
        
        # Initial market state
        current_shares = {segment.segment_id: segment.current_market_share 
                        for segment in self.market_segments}
        
        # Simulate month by month
        for month in range(time_horizon):
            monthly_response = {
                'month': month + 1,
                'market_shares': current_shares.copy(),
                'intervention_impacts': {},
                'total_market_size': sum(segment.size for segment in self.market_segments)
            }
            
            # Apply intervention impacts
            for intervention in interventions:
                if month >= intervention.implementation_time:
                    # Calculate intervention success
                    success = np.random.random() < intervention.success_probability
                    
                    if success:
                        # Apply intervention to target segments
                        for target_segment_id in intervention.target_segments:
                            if target_segment_id in current_shares:
                                # Calculate market share change
                                impact = intervention.expected_impact.get('market_share', 0.0)
                                current_shares[target_segment_id] += impact * 0.01  # Monthly impact
                                current_shares[target_segment_id] = min(1.0, current_shares[target_segment_id])
                                
                                monthly_response['intervention_impacts'][target_segment_id] = impact
                    
                    # Track successful interventions
                    if success and intervention.intervention_id not in simulation_results['successful_interventions']:
                        simulation_results['successful_interventions'].append(intervention.intervention_id)
            
            simulation_results['monthly_responses'].append(monthly_response)
        
        # Calculate final results
        simulation_results['final_market_share'] = current_shares
        
        # Calculate total ROI
        total_investment = sum(intervention.investment_cost for intervention in interventions)
        total_revenue = sum(current_shares.values()) * 1000  # Simplified revenue calculation
        simulation_results['total_roi'] = (total_revenue - total_investment) / total_investment
        
        # Calculate market growth
        initial_total = sum(segment.current_market_share for segment in self.market_segments)
        final_total = sum(current_shares.values())
        simulation_results['market_growth'] = (final_total - initial_total) / initial_total
        
        return simulation_results
    
    def optimize_market_strategy(self, budget: float, 
                                 time_horizon: int = 12) -> Dict[str, Any]:
        """Optimize market strategy within budget constraints"""
        
        # Generate all possible intervention combinations
        best_strategy = {
            'selected_interventions': [],
            'expected_roi': 0.0,
            'total_cost': 0.0,
            'market_impact': 0.0
        }
        
        # Try different intervention combinations
        for r in range(1, min(len(self.interventions), 4) + 1):
            for combination in self._generate_combinations(self.interventions, r):
                total_cost = sum(intervention.investment_cost for intervention in combination)
                
                if total_cost <= budget:
                    # Simulate this combination
                    simulation_result = self.simulate_market_response(combination, time_horizon)
                    
                    # Evaluate strategy
                    strategy_score = simulation_result['total_roi'] + simulation_result['market_growth'] * 0.5
                    
                    if strategy_score > best_strategy['expected_roi']:
                        best_strategy = {
                            'selected_interventions': combination,
                            'expected_roi': strategy_score,
                            'total_cost': total_cost,
                            'market_impact': simulation_result['market_growth'],
                            'simulation_result': simulation_result
                        }
        
        return best_strategy
    
    def _generate_combinations(self, interventions: List[StrategicIntervention], 
                              r: int) -> List[List[StrategicIntervention]]:
        """Generate combinations of interventions"""
        
        combinations = []
        
        def backtrack(start, current_combination):
            if len(current_combination) == r:
                combinations.append(current_combination.copy())
                return
            
            for i in range(start, len(interventions)):
                current_combination.append(interventions[i])
                backtrack(i + 1, current_combination)
                current_combination.pop()
        
        backtrack(0, [])
        return combinations
    
    def generate_market_report(self) -> Dict[str, Any]:
        """Generate comprehensive market analysis report"""
        
        opportunities = self.analyze_market_opportunities()
        
        report = {
            'report_timestamp': time.time(),
            'market_overview': {
                'total_segments': len(self.market_segments),
                'total_market_size': opportunities['market_size'],
                'growth_potential': opportunities['growth_potential'],
                'current_market_position': sum(segment.current_market_share for segment in self.market_segments)
            },
            'segment_analysis': {},
            'strategic_opportunities': opportunities,
            'recommended_actions': [],
            'performance_metrics': {}
        }
        
        # Segment analysis
        for segment in self.market_segments:
            report['segment_analysis'][segment.segment_id] = {
                'type': segment.segment_type.value,
                'size': segment.size,
                'market_share': segment.current_market_share,
                'growth_potential': segment.growth_potential,
                'price_elasticity': segment.price_elasticity,
                'quality_sensitivity': segment.quality_sensitivity,
                'convenience_sensitivity': segment.convenience_sensitivity,
                'sustainability_sensitivity': segment.sustainability_sensitivity
            }
        
        # Recommended actions
        if opportunities['high_growth_segments']:
            report['recommended_actions'].append('Focus on high-growth segments for expansion')
        
        if opportunities['underserved_segments']:
            report['recommended_actions'].append('Develop strategies for underserved market segments')
        
        if 'quality_leadership' in opportunities['competitive_advantages']:
            report['recommended_actions'].append('Leverage quality leadership for competitive advantage')
        
        if 'sustainability_focus' in opportunities['competitive_advantages']:
            report['recommended_actions'].append('Emphasize sustainability in market positioning')
        
        report['recommended_actions'].append('Implement strategic interventions to shape market demand')
        report['recommended_actions'].append('Monitor market response and adjust strategy accordingly')
        
        return report
```

```python
def visualize_market_shaping(market_shaper: MarketShaper, 
                            opportunities: Dict[str, Any],
                            optimal_strategy: Dict[str, Any],
                            simulation_result: Dict[str, Any]):
    """Visualize market shaping results"""
    
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
    
    # Plot 1: Market segment analysis
    ax1.set_title('Market Segment Analysis', fontsize=14, fontweight='bold')
    
    segment_names = [segment.segment_id for segment in market_shaper.market_segments]
    segment_sizes = [segment.size for segment in market_shaper.market_segments]
    market_shares = [segment.current_market_share for segment in market_shaper.market_segments]
    growth_potentials = [segment.growth_potential for segment in market_shaper.market_segments]
    
    x = np.arange(len(segment_names))
    width = 0.25
    
    bars1 = ax1.bar(x - width, segment_sizes, width, label='Size', alpha=0.7, color='lightblue')
    bars2 = ax1.bar(x, [s * 1000 for s in market_shares], width, label='Market Share', alpha=0.7, color='lightcoral')
    bars3 = ax1.bar(x + width, [g * 1000 for g in growth_potentials], width, label='Growth Potential', alpha=0.7, color='lightgreen')
    
    ax1.set_xlabel('Market Segment')
    ax1.set_ylabel('Customers')
    ax1.set_xticks(x)
    ax1.set_xticklabels([name.replace('_', ' ').title() for name in segment_names], rotation=45)
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Plot 2: Intervention ROI comparison
    ax2.set_title('Strategic Intervention ROI Analysis', fontsize=14, fontweight='bold')
    
    intervention_names = [intervention.intervention_id for intervention in market_shaper.interventions]
    intervention_costs = [intervention.investment_cost for intervention in market_shaper.interventions]
    
    # Calculate ROI for each intervention
    intervention_rois = []
    for intervention in market_shaper.interventions:
        # Simplified ROI calculation
        expected_impact = sum(intervention.expected_impact.values())
        roi = (expected_impact * 1000 - intervention.investment_cost) / intervention.investment_cost
        intervention_rois.append(roi)
    
    x = np.arange(len(intervention_names))
    width = 0.35
    
    bars1 = ax2.bar(x - width/2, intervention_costs, width, label='Investment Cost', alpha=0.7, color='orange')
    bars2 = ax2.bar(x + width/2, [r * 1000 for r in intervention_rois], width, label='Expected Return', alpha=0.7, color='green')
    
    ax2.set_xlabel('Intervention')
    ax2.set_ylabel('Amount ($)')
    ax2.set_xticks(x)
    ax2.set_xticklabels([name.replace('_', ' ').title() for name in intervention_names], rotation=45)
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # Plot 3: Market share evolution
    ax3.set_title('Market Share Evolution Over Time', fontsize=14, fontweight='bold')
    
    months = range(1, len(simulation_result['monthly_responses']) + 1)
    
    # Plot market share evolution for each segment
    for segment in market_shaper.market_segments:
        share_evolution = [response['market_shares'].get(segment.segment_id, 0) 
                          for response in simulation_result['monthly_responses']]
        ax3.plot(months, share_evolution, marker='o', label=segment.segment_id.replace('_', ' ').title(), linewidth=2)
    
    ax3.set_xlabel('Month')
    ax3.set_ylabel('Market Share')
    ax3.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    ax3.grid(True, alpha=0.3)
    
    # Plot 4: Strategic impact visualization
    ax4.set_title('Strategic Impact Assessment', fontsize=14, fontweight='bold')
    
    # Create impact metrics
    impact_metrics = {
        'Market Growth': simulation_result['market_growth'] * 100,
        'ROI': simulation_result['total_roi'] * 100,
        'Success Rate': len(simulation_result['successful_interventions']) / len(optimal_strategy['selected_interventions']) * 100,
        'Budget Utilization': optimal_strategy['total_cost'] / budget * 100
    }
    
    metric_names = list(impact_metrics.keys())
    metric_values = list(impact_metrics.values())
    
    colors = ['lightblue', 'lightgreen', 'lightcoral', 'lightyellow']
    
    bars = ax4.bar(metric_names, metric_values, color=colors, alpha=0.8)
    
    ax4.set_ylabel('Percentage (%)')
    ax4.set_ylim(0, 100)
    ax4.grid(True, alpha=0.3)
    
    # Add value labels on bars
    for bar, value in zip(bars, metric_values):
        height = bar.get_height()
        ax4.text(bar.get_x() + bar.get_width()/2., height + 1,
                f'{value:.1f}%', ha='center', va='bottom', fontsize=10)
    
    plt.tight_layout()
    plt.show()
    
    # Print detailed analysis
    print("\nStrategic Impact Analysis:")
    print("=" * 30)
    
    total_market_size = sum(segment.size for segment in market_shaper.market_segments)
    initial_market_value = sum(segment.current_market_share * segment.size for segment in market_shaper.market_segments)
    final_market_value = sum(simulation_result['final_market_share'][segment.segment_id] * segment.size for segment in market_shaper.market_segments)
    
    print(f"Market Value Analysis:")
    print(f"  Initial Market Value: {initial_market_value:.0f} customers")
    print(f"  Final Market Value: {final_market_value:.0f} customers")
    print(f"  Value Growth: {final_market_value - initial_market_value:.0f} customers")
    print(f"  Growth Rate: {(final_market_value / initial_market_value - 1) * 100:.1f}%")
    
    print(f"\nInvestment Analysis:")
    print(f"  Total Investment: ${optimal_strategy['total_cost']:,}")
    print(f"  Total ROI: {simulation_result['total_roi']:.2f}")
    print(f"  Payback Period: {optimal_strategy['total_cost'] / (final_market_value * 100):.2f} years")
    
    print(f"\nCompetitive Position:")
    print(f"  Market Share Growth: {simulation_result['market_growth']:.1%}")
    print(f"  Competitive Advantage: {'Strong' if simulation_result['market_growth'] > 0.1 else 'Moderate' if simulation_result['market_growth'] > 0.05 else 'Limited'}")
    print(f"  Strategic Position: {'Market Leader' if simulation_result['market_growth'] > 0.15 else 'Strong Contender' if simulation_result['market_growth'] > 0.1 else 'Growing Player'}")
    
    return simulation_result

# Visualize market shaping results
visualize_market_shaping(market_shaper, opportunities, optimal_strategy, simulation_result)
```
