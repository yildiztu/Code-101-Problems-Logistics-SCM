```python
from typing import Tuple, List, Dict, Set

# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass, field
from typing import List, Tuple, Optional, Dict, Any
import random
from collections import defaultdict, deque
import warnings
from enum import Enum
import time
warnings.filterwarnings('ignore')

# Set style for better visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)
```

```python
class AgentType(Enum):
    """Types of AI agents in the metaverse"""
    COORDINATOR = "coordinator"
    OPERATOR = "operator"
    OPTIMIZER = "optimizer"
    ANALYST = "analyst"
    INTERFACE = "interface"

@dataclass
class MetaverseSensor:
    """IoT sensor in the metaverse environment"""
    sensor_id: str
    location: Tuple[float, float, float]  # 3D coordinates
    sensor_type: str  # 'temperature', 'humidity', 'motion', 'inventory', 'traffic'
    value: float = 0.0
    last_update: float = 0.0
    accuracy: float = 0.95
    status: str = "active"  # 'active', 'maintenance', 'offline'
    
    def update_value(self, new_value: float, timestamp: float):
        """Update sensor value with timestamp"""
        self.value = new_value
        self.last_update = timestamp
        
        # Add realistic noise
        noise = np.random.normal(0, 0.05)
        self.value += noise

@dataclass
class AIAgent:
    """AI agent operating in the metaverse"""
    agent_id: str
    agent_type: AgentType
    capabilities: List[str] = field(default_factory=list)
    location: Tuple[float, float, float] = (0.0, 0.0, 0.0)
    status: str = "active"
    performance_score: float = 0.8
    decision_history: List[Dict[str, Any]] = field(default_factory=list)
    learning_rate: float = 0.1
    
    def make_decision(self, context: Dict[str, Any]) -> Dict[str, Any]:
        """Make autonomous decision based on context"""
        
        decision = {
            'agent_id': self.agent_id,
            'timestamp': time.time(),
            'decision_type': self.agent_type.value,
            'action': None,
            'confidence': 0.0,
            'reasoning': ''
        }
        
        # Agent-specific decision logic
        if self.agent_type == AgentType.COORDINATOR:
            decision = self._coordinate_decision(context)
        elif self.agent_type == AgentType.OPERATOR:
            decision = self._operate_decision(context)
        elif self.agent_type == AgentType.OPTIMIZER:
            decision = self._optimize_decision(context)
        elif self.agent_type == AgentType.ANALYST:
            decision = self._analyze_decision(context)
        elif self.agent_type == AgentType.INTERFACE:
            decision = self._interface_decision(context)
        
        self.decision_history.append(decision)
        return decision
    
    def _coordinate_decision(self, context: Dict[str, Any]) -> Dict[str, Any]:
        """Coordinator agent decision logic"""
        
        # Analyze system status
        system_status = context.get('system_status', 'normal')
        agent_status = context.get('agent_status', {})
        
        if system_status == 'overloaded':
            action = 'rebalance_load'
            confidence = 0.9
            reasoning = 'System overloaded - redistributing tasks among agents'
        elif system_status == 'underutilized':
            action = 'consolidate_resources'
            confidence = 0.8
            reasoning = 'System underutilized - consolidating resources for efficiency'
        else:
            action = 'maintain_balance'
            confidence = 0.7
            reasoning = 'System normal - maintaining current balance'
        
        return {
            'agent_id': self.agent_id,
            'timestamp': time.time(),
            'decision_type': self.agent_type.value,
            'action': action,
            'confidence': confidence,
            'reasoning': reasoning
        }
    
    def _operate_decision(self, context: Dict[str, Any]) -> Dict[str, Any]:
        """Operator agent decision logic"""
        
        task_queue = context.get('task_queue', [])
        resource_status = context.get('resource_status', {})
        
        if len(task_queue) > 10:
            action = 'increase_capacity'
            confidence = 0.8
            reasoning = 'High task queue - increasing operational capacity'
        elif len(task_queue) < 3:
            action = 'optimize_efficiency'
            confidence = 0.7
            reasoning = 'Low task queue - optimizing operational efficiency'
        else:
            action = 'maintain_current'
            confidence = 0.6
            reasoning = 'Normal task load - maintaining current operations'
        
        return {
            'agent_id': self.agent_id,
            'timestamp': time.time(),
            'decision_type': self.agent_type.value,
            'action': action,
            'confidence': confidence,
            'reasoning': reasoning
        }
    
    def _optimize_decision(self, context: Dict[str, Any]) -> Dict[str, Any]:
        """Optimizer agent decision logic"""
        
        performance_metrics = context.get('performance_metrics', {})
        efficiency = performance_metrics.get('efficiency', 0.8)
        
        if efficiency < 0.7:
            action = 'reconfigure_network'
            confidence = 0.9
            reasoning = 'Low efficiency detected - reconfiguring network topology'
        elif efficiency > 0.9:
            action = 'fine_tune_parameters'
            confidence = 0.7
            reasoning = 'High efficiency - fine-tuning parameters for optimization'
        else:
            action = 'monitor_performance'
            confidence = 0.6
            reasoning = 'Normal efficiency - monitoring performance metrics'
        
        return {
            'agent_id': self.agent_id,
            'timestamp': time.time(),
            'decision_type': self.agent_type.value,
            'action': action,
            'confidence': confidence,
            'reasoning': reasoning
        }
    
    def _analyze_decision(self, context: Dict[str, Any]) -> Dict[str, Any]:
        """Analyst agent decision logic"""
        
        data_volume = context.get('data_volume', 1000)
        anomaly_count = context.get('anomaly_count', 0)
        
        if anomaly_count > 10:
            action = 'investigate_anomalies'
            confidence = 0.9
            reasoning = 'High anomaly count - investigating potential issues'
        elif data_volume > 5000:
            action = 'analyze_patterns'
            confidence = 0.8
            reasoning = 'High data volume - analyzing patterns for insights'
        else:
            action = 'generate_report'
            confidence = 0.7
            reasoning = 'Normal conditions - generating analytical report'
        
        return {
            'agent_id': self.agent_id,
            'timestamp': time.time(),
            'decision_type': self.agent_type.value,
            'action': action,
            'confidence': confidence,
            'reasoning': reasoning
        }
    
    def _interface_decision(self, context: Dict[str, Any]) -> Dict[str, Any]:
        """Interface agent decision logic"""
        
        user_activity = context.get('user_activity', 'low')
        interface_load = context.get('interface_load', 0.5)
        
        if user_activity == 'high':
            action = 'enhance_interface'
            confidence = 0.8
            reasoning = 'High user activity - enhancing interface responsiveness'
        elif interface_load > 0.8:
            action = 'optimize_rendering'
            confidence = 0.7
            reasoning = 'High interface load - optimizing rendering performance'
        else:
            action = 'maintain_interface'
            confidence = 0.6
            reasoning = 'Normal conditions - maintaining current interface'
        
        return {
            'agent_id': self.agent_id,
            'timestamp': time.time(),
            'decision_type': self.agent_type.value,
            'action': action,
            'confidence': confidence,
            'reasoning': reasoning
        }
    
    def update_performance(self, outcome: float):
        """Update agent performance based on outcome"""
        self.performance_score = (1 - self.learning_rate) * self.performance_score + self.learning_rate * outcome

@dataclass
class MetaverseEnvironment:
    """Physical-cyber synthesis metaverse environment"""
    environment_id: str
    sensors: List[MetaverseSensor] = field(default_factory=list)
    agents: List[AIAgent] = field(default_factory=list)
    physical_state: Dict[str, Any] = field(default_factory=dict)
    virtual_state: Dict[str, Any] = field(default_factory=dict)
    synchronization_status: float = 1.0
    performance_metrics: Dict[str, float] = field(default_factory=dict)
    
    def add_sensor(self, sensor: MetaverseSensor):
        """Add a sensor to the metaverse"""
        self.sensors.append(sensor)
    
    def add_agent(self, agent: AIAgent):
        """Add an AI agent to the metaverse"""
        self.agents.append(agent)
    
    def update_sensors(self, timestamp: float):
        """Update all sensor readings"""
        
        for sensor in self.sensors:
            if sensor.status == 'active':
                # Generate realistic sensor values
                if sensor.sensor_type == 'temperature':
                    base_value = 20 + 10 * np.sin(timestamp / 3600 * 2 * np.pi)
                    new_value = base_value + np.random.normal(0, 2)
                elif sensor.sensor_type == 'humidity':
                    base_value = 50 + 20 * np.sin(timestamp / 3600 * 2 * np.pi + np.pi/4)
                    new_value = base_value + np.random.normal(0, 5)
                elif sensor.sensor_type == 'motion':
                    new_value = np.random.uniform(0, 100)  # Motion activity level
                elif sensor.sensor_type == 'inventory':
                    new_value = np.random.uniform(0, 1000)  # Inventory level
                elif sensor.sensor_type == 'traffic':
                    new_value = np.random.uniform(0, 50)  # Traffic flow
                else:
                    new_value = np.random.uniform(0, 100)
                
                sensor.update_value(new_value, timestamp)
    
    def synchronize_states(self) -> float:
        """Synchronize physical and virtual states"""
        
        # Calculate synchronization quality
        active_sensors = [s for s in self.sensors if s.status == 'active']
        
        if not active_sensors:
            return 0.0
        
        # Calculate average sensor accuracy
        avg_accuracy = np.mean([s.accuracy for s in active_sensors])
        
        # Calculate state difference
        state_diff = 0.0
        for sensor in active_sensors:
            # Simulate state difference based on sensor accuracy
            state_diff += (1 - sensor.accuracy) * 0.1
        
        self.synchronization_status = avg_accuracy - (state_diff / len(active_sensors))
        
        return self.synchronization_status
    
    def run_agent_decisions(self) -> List[Dict[str, Any]]:
        """Run autonomous agent decision making"""
        
        decisions = []
        
        # Create context for agents
        context = {
            'system_status': self._calculate_system_status(),
            'agent_status': {agent.agent_id: agent.status for agent in self.agents},
            'task_queue': self._get_task_queue(),
            'resource_status': self._get_resource_status(),
            'performance_metrics': self.performance_metrics,
            'data_volume': len(self.sensors) * 100,  # Simplified data volume
            'anomaly_count': self._count_anomalies(),
            'user_activity': self._get_user_activity(),
            'interface_load': self._get_interface_load()
        }
        
        # Each agent makes decision
        for agent in self.agents:
            if agent.status == 'active':
                decision = agent.make_decision(context)
                decisions.append(decision)
        
        return decisions
    
    def _calculate_system_status(self) -> str:
        """Calculate overall system status"""
        
        active_agents = len([a for a in self.agents if a.status == 'active'])
        total_agents = len(self.agents)
        
        if total_agents == 0:
            return 'unknown'
        
        agent_ratio = active_agents / total_agents
        
        if agent_ratio > 0.8:
            return 'normal'
        elif agent_ratio > 0.5:
            return 'degraded'
        else:
            return 'critical'
    
    def _get_task_queue(self) -> List[str]:
        """Get current task queue"""
        # Simulate task queue based on sensor activity
        active_sensors = [s for s in self.sensors if s.status == 'active']
        task_count = int(len(active_sensors) * 0.5)
        return [f'task_{i}' for i in range(task_count)]
    
    def _get_resource_status(self) -> Dict[str, float]:
        """Get resource status"""
        return {
            'cpu_usage': np.random.uniform(0.3, 0.8),
            'memory_usage': np.random.uniform(0.4, 0.7),
            'network_usage': np.random.uniform(0.2, 0.6),
            'storage_usage': np.random.uniform(0.5, 0.9)
        }
    
    def _count_anomalies(self) -> int:
        """Count system anomalies"""
        # Simulate anomaly detection based on sensor readings
        anomaly_count = 0
        
        for sensor in self.sensors:
            if sensor.status == 'active':
                # Simple anomaly detection
                if sensor.sensor_type == 'temperature' and (sensor.value < 10 or sensor.value > 35):
                    anomaly_count += 1
                elif sensor.sensor_type == 'humidity' and (sensor.value < 20 or sensor.value > 80):
                    anomaly_count += 1
        
        return anomaly_count
    
    def _get_user_activity(self) -> str:
        """Get current user activity level"""
        activity_level = np.random.random()
        
        if activity_level > 0.7:
            return 'high'
        elif activity_level > 0.3:
            return 'medium'
        else:
            return 'low'
    
    def _get_interface_load(self) -> float:
        """Get interface load"""
        return np.random.uniform(0.2, 0.9)
    
    def update_performance_metrics(self):
        """Update performance metrics"""
        
        # Calculate efficiency
        active_agents = [a for a in self.agents if a.status == 'active']
        if active_agents:
            avg_performance = np.mean([a.performance_score for a in active_agents])
        else:
            avg_performance = 0.0
        
        self.performance_metrics = {
            'efficiency': avg_performance,
            'synchronization': self.synchronization_status,
            'sensor_coverage': len([s for s in self.sensors if s.status == 'active']) / len(self.sensors) if self.sensors else 0,
            'agent_utilization': len(active_agents) / len(self.agents) if self.agents else 0,
            'data_quality': np.mean([s.accuracy for s in self.sensors if s.status == 'active']) if self.sensors else 0,
            'response_time': np.random.uniform(0.1, 2.0),  # seconds
            'throughput': np.random.uniform(100, 1000),  # operations per second
        }
    
    def generate_metaverse_report(self) -> Dict[str, Any]:
        """Generate comprehensive metaverse report"""
        
        report = {
            'report_timestamp': time.time(),
            'environment_id': self.environment_id,
            'system_overview': {
                'total_sensors': len(self.sensors),
                'active_sensors': len([s for s in self.sensors if s.status == 'active']),
                'total_agents': len(self.agents),
                'active_agents': len([a for a in self.agents if a.status == 'active']),
                'synchronization_status': self.synchronization_status,
                'system_status': self._calculate_system_status()
            },
            'sensor_analysis': {},
            'agent_analysis': {},
            'performance_metrics': self.performance_metrics,
            'recommendations': []
        }
        
        # Sensor analysis
        sensor_types = defaultdict(int)
        for sensor in self.sensors:
            sensor_types[sensor.sensor_type] += 1
        
        report['sensor_analysis'] = {
            'sensor_types': dict(sensor_types),
            'average_accuracy': np.mean([s.accuracy for s in self.sensors if s.status == 'active']) if self.sensors else 0,
            'coverage_area': len(self.sensors) * 100,  # Simplified coverage
        }
        
        # Agent analysis
        agent_types = defaultdict(int)
        for agent in self.agents:
            agent_types[agent.agent_type.value] += 1
        
        report['agent_analysis'] = {
            'agent_types': dict(agent_types),
            'average_performance': np.mean([a.performance_score for a in self.agents]) if self.agents else 0,
            'decision_frequency': len([d for a in self.agents for d in a.decision_history]) / len(self.agents) if self.agents else 0,
        }
        
        # Generate recommendations
        if self.synchronization_status < 0.8:
            report['recommendations'].append('Improve physical-cyber synchronization')
        
        if len([s for s in self.sensors if s.status == 'active']) < len(self.sensors) * 0.8:
            report['recommendations'].append('Maintain sensor health and availability')
        
        if self.performance_metrics.get('efficiency', 0) < 0.7:
            report['recommendations'].append('Optimize agent performance and coordination')
        
        return report
```

```python
try:
    # Initialize Metaverse Environment
    print("Initializing Physical-Cyber Synthesis Metaverse...")
    print("=" * 50)
    
    metaverse = MetaverseEnvironment(environment_id="warehouse_metaverse_001")
    
    # Create sensors (1000+ sensors for comprehensive coverage)
    sensor_types = ['temperature', 'humidity', 'motion', 'inventory', 'traffic']
    num_sensors_per_type = 200  # 200 sensors per type = 1000 total
    
    print(f"Creating {len(sensor_types) * num_sensors_per_type} sensors...")
    
    for sensor_type in sensor_types:
        for i in range(num_sensors_per_type):
            sensor = MetaverseSensor(
                sensor_id=f"{sensor_type}_{i}",
                location=(np.random.uniform(0, 100), np.random.uniform(0, 100), np.random.uniform(0, 10)),
                sensor_type=sensor_type,
                accuracy=np.random.uniform(0.9, 0.99)
            )
            metaverse.add_sensor(sensor)
    
    print(f"Sensors Created: {len(metaverse.sensors)}")
    for sensor_type in sensor_types:
        count = len([s for s in metaverse.sensors if s.sensor_type == sensor_type])
        print(f"  {sensor_type.title()}: {count} sensors")
    
    # Create AI agents
    agent_configurations = [
        (AgentType.COORDINATOR, 5),
        (AgentType.OPERATOR, 10),
        (AgentType.OPTIMIZER, 8),
        (AgentType.ANALYST, 6),
        (AgentType.INTERFACE, 4)
    ]
    
    print(f"\nCreating AI agents...")
    
    for agent_type, count in agent_configurations:
        for i in range(count):
            agent = AIAgent(
                agent_id=f"{agent_type.value}_{i}",
                agent_type=agent_type,
                capabilities=[f"{agent_type.value}_capability_{j}" for j in range(3)],
                location=(np.random.uniform(0, 100), np.random.uniform(0, 100), np.random.uniform(0, 10)),
                performance_score=np.random.uniform(0.7, 0.95)
            )
            metaverse.add_agent(agent)
    
    print(f"AI Agents Created: {len(metaverse.agents)}")
    for agent_type, count in agent_configurations:
        print(f"  {agent_type.value.title()}: {count} agents")
    
    # Initialize performance metrics
    metaverse.update_performance_metrics()
    
    print(f"\nInitial Performance Metrics:")
    for metric, value in metaverse.performance_metrics.items():
        print(f"  {metric.replace('_', ' ').title()}: {value:.3f}")
    
    print(f"\nMetaverse Environment Ready:")
    print(f"  Environment ID: {metaverse.environment_id}")
    print(f"  Total Sensors: {len(metaverse.sensors)}")
    print(f"  Total Agents: {len(metaverse.agents)}")
    print(f"  Synchronization Status: {metaverse.synchronization_status:.3f}")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    try:
        def run_metaverse_simulation(metaverse: MetaverseEnvironment, duration_minutes: int = 5):
            """Run metaverse simulation with real-time updates"""
            
            print(f"Running Metaverse Simulation for {duration_minutes} minutes...")
            print("=" * 60)
            
            simulation_data = {
                'timestamps': [],
                'synchronization_status': [],
                'active_sensors': [],
                'active_agents': [],
                'performance_metrics': [],
                'agent_decisions': [],
                'anomaly_counts': []
            }
            
            start_time = time.time()
            
            for minute in range(duration_minutes):
                for second in range(60):
                    current_time = time.time()
                    
                    # Update sensors
                    metaverse.update_sensors(current_time)
                    
                    # Synchronize states
                    sync_status = metaverse.synchronize_states()
                    
                    # Run agent decisions
                    decisions = metaverse.run_agent_decisions()
                    
                    # Update performance metrics
                    metaverse.update_performance_metrics()
                    
                    # Collect data every second
                    simulation_data['timestamps'].append(current_time)
                    simulation_data['synchronization_status'].append(sync_status)
                    simulation_data['active_sensors'].append(len([s for s in metaverse.sensors if s.status == 'active']))
                    simulation_data['active_agents'].append(len([a for a in metaverse.agents if a.status == 'active']))
                    simulation_data['performance_metrics'].append(metaverse.performance_metrics.copy())
                    simulation_data['agent_decisions'].append(decisions)
                    simulation_data['anomaly_counts'].append(metaverse._count_anomalies())
                    
                    # Print status every minute
                    if second == 0:
                        print(f"Minute {minute + 1}: Sync={sync_status:.3f}, Sensors={len([s for s in metaverse.sensors if s.status == 'active'])}, Agents={len([a for a in metaverse.agents if a.status == 'active'])}")
                        
                        # Show sample decisions
                        if decisions:
                            print(f"  Sample Decisions: {len(decisions)} total")
                            for i, decision in enumerate(decisions[:3]):
                                print(f"    {decision['agent_id']}: {decision['action']} (confidence: {decision['confidence']:.2f})")
                    
                    time.sleep(0.01)  # Small delay for realistic simulation
            
            end_time = time.time()
            
            print(f"\nSimulation Summary:")
            print(f"Duration: {end_time - start_time:.2f} seconds")
            print(f"Total Updates: {len(simulation_data['timestamps'])}")
            print(f"Average Synchronization: {np.mean(simulation_data['synchronization_status']):.3f}")
            print(f"Average Active Sensors: {np.mean(simulation_data['active_sensors']):.1f}")
            print(f"Average Active Agents: {np.mean(simulation_data['active_agents']):.1f}")
            print(f"Total Decisions Made: {sum(len(decisions) for decisions in simulation_data['agent_decisions'])}")
            print(f"Average Anomalies: {np.mean(simulation_data['anomaly_counts']):.1f}")
            
            return simulation_data
        
        # Run the simulation
        simulation_data = run_metaverse_simulation(metaverse, duration_minutes=3)
        
        # Generate final report
        final_report = metaverse.generate_metaverse_report()
        
        print(f"\nFinal Metaverse Report:")
        print("=" * 30)
        
        overview = final_report['system_overview']
        print(f"System Overview:")
        print(f"  Total Sensors: {overview['total_sensors']}")
        print(f"  Active Sensors: {overview['active_sensors']} ({overview['active_sensors']/overview['total_sensors']:.1%})")
        print(f"  Total Agents: {overview['total_agents']}")
        print(f"  Active Agents: {overview['active_agents']} ({overview['active_agents']/overview['total_agents']:.1%})")
        print(f"  Synchronization Status: {overview['synchronization_status']:.3f}")
        print(f"  System Status: {overview['system_status']}")
        
        print(f"\nRecommendations:")
        for i, recommendation in enumerate(final_report['recommendations'], 1):
            print(f"  {i}. {recommendation}")
    except Exception as e:
        print(f'Error in cell: {e}')
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    def visualize_metaverse_performance(metaverse: MetaverseEnvironment, 
                                       simulation_data: Dict[str, List]):
        """Visualize metaverse performance and analytics"""
        
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
        
        # Plot 1: Synchronization status over time
        ax1.set_title('Physical-Cyber Synchronization Status', fontsize=14, fontweight='bold')
        
        time_minutes = [(t - simulation_data['timestamps'][0]) / 60 for t in simulation_data['timestamps']]
        
        ax1.plot(time_minutes, simulation_data['synchronization_status'], 'b-', linewidth=2)
        ax1.axhline(y=0.8, color='r', linestyle='--', alpha=0.7, label='Good Sync Threshold')
        ax1.axhline(y=0.9, color='g', linestyle='--', alpha=0.7, label='Excellent Sync Threshold')
        ax1.set_xlabel('Time (minutes)')
        ax1.set_ylabel('Synchronization Status (0-1)')
        ax1.set_ylim(0, 1)
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # Plot 2: Active sensors and agents
        ax2.set_title('System Resource Utilization', fontsize=14, fontweight='bold')
        
        ax2_twin = ax2.twinx()
        
        line1 = ax2.plot(time_minutes, simulation_data['active_sensors'], 'b-', label='Active Sensors', linewidth=2)
        ax2.set_xlabel('Time (minutes)')
        ax2.set_ylabel('Active Sensors', color='b')
        ax2.tick_params(axis='y', labelcolor='b')
        
        line2 = ax2_twin.plot(time_minutes, simulation_data['active_agents'], 'r-', label='Active Agents', linewidth=2)
        ax2_twin.set_ylabel('Active Agents', color='r')
        ax2_twin.tick_params(axis='y', labelcolor='r')
        
        ax2.legend([line1, line2], ['Active Sensors', 'Active Agents'], loc='upper left')
        ax2.grid(True, alpha=0.3)
        
        # Plot 3: Performance metrics evolution
        ax3.set_title('Performance Metrics Evolution', fontsize=14, fontweight='bold')
        
        # Extract key performance metrics
        efficiency_scores = [metrics.get('efficiency', 0) for metrics in simulation_data['performance_metrics']]
        data_quality_scores = [metrics.get('data_quality', 0) for metrics in simulation_data['performance_metrics']]
        
        ax3.plot(time_minutes, efficiency_scores, 'g-', label='Efficiency', linewidth=2)
        ax3.plot(time_minutes, data_quality_scores, 'orange', label='Data Quality', linewidth=2)
        ax3.set_xlabel('Time (minutes)')
        ax3.set_ylabel('Performance Score (0-1)')
        ax3.set_ylim(0, 1)
        ax3.legend()
        ax3.grid(True, alpha=0.3)
        
        # Plot 4: Agent decision analysis
        ax4.set_title('Agent Decision Analysis', fontsize=14, fontweight='bold')
        
        # Count decisions by type
        decision_types = defaultdict(int)
        for decisions in simulation_data['agent_decisions']:
            for decision in decisions:
                decision_types[decision['decision_type']] += 1
        
        if decision_types:
            types = list(decision_types.keys())
            counts = list(decision_types.values())
            
            bars = ax4.bar(types, counts, alpha=0.7, color='lightblue')
            ax4.set_xlabel('Agent Type')
            ax4.set_ylabel('Number of Decisions')
            ax4.set_xticklabels(types, rotation=45)
            ax4.grid(True, alpha=0.3)
            
            # Add value labels
            for bar, count in zip(bars, counts):
                height = bar.get_height()
                ax4.text(bar.get_x() + bar.get_width()/2., height + max(counts) * 0.01,
                        str(count), ha='center', va='bottom', fontsize=10)
        
        plt.tight_layout()
        plt.show()
        
        # Additional analysis
        print("\nMetaverse Performance Analysis:")
        print("=" * 35)
        
        # Calculate system health score
        avg_sync = np.mean(simulation_data['synchronization_status'])
        avg_efficiency = np.mean([metrics.get('efficiency', 0) for metrics in simulation_data['performance_metrics']])
        avg_data_quality = np.mean([metrics.get('data_quality', 0) for metrics in simulation_data['performance_metrics']])
        
        system_health = (avg_sync + avg_efficiency + avg_data_quality) / 3
        
        print(f"System Health Score: {system_health:.3f}")
        print(f"  Synchronization: {avg_sync:.3f}")
        print(f"  Efficiency: {avg_efficiency:.3f}")
        print(f"  Data Quality: {avg_data_quality:.3f}")
        
        # Agent performance analysis
        agent_performance = {}
        for agent in metaverse.agents:
            agent_performance[agent.agent_id] = agent.performance_score
        
        if agent_performance:
            best_agent = max(agent_performance, key=agent_performance.get)
            worst_agent = min(agent_performance, key=agent_performance.get)
            
            print(f"\nAgent Performance Analysis:")
            print(f"  Best Agent: {best_agent} ({agent_performance[best_agent]:.3f})")
            print(f"  Worst Agent: {worst_agent} ({agent_performance[worst_agent]:.3f})")
            print(f"  Average Performance: {np.mean(list(agent_performance.values())):.3f}")
        
        # Anomaly analysis
        anomaly_counts = simulation_data['anomaly_counts']
        avg_anomalies = np.mean(anomaly_counts)
        max_anomalies = max(anomaly_counts)
        
        print(f"\nAnomaly Analysis:")
        print(f"  Average Anomalies: {avg_anomalies:.1f}")
        print(f"  Peak Anomalies: {max_anomalies}")
        print(f"  Anomaly Rate: {avg_anomalies / len(metaverse.sensors) * 100:.2f}%")
        
        # Decision analysis
        total_decisions = sum(len(decisions) for decisions in simulation_data['agent_decisions'])
        decisions_per_minute = total_decisions / len(simulation_data['timestamps']) * 60
        
        print(f"\nDecision Analysis:")
        print(f"  Total Decisions: {total_decisions}")
        print(f"  Decisions per Minute: {decisions_per_minute:.1f}")
        print(f"  Decision Rate: {decisions_per_minute / len(metaverse.agents):.2f} per agent per minute")
        
        # Assessment
        print(f"\nOverall Assessment:")
        if system_health > 0.85:
            print(f"Status: EXCELLENT - Metaverse operating at peak performance")
        elif system_health > 0.75:
            print(f"Status: GOOD - Metaverse performing well with minor issues")
        elif system_health > 0.65:
            print(f"Status: ACCEPTABLE - Metaverse functional but needs optimization")
        else:
            print(f"Status: NEEDS IMPROVEMENT - Metaverse requires significant optimization")
        
        return simulation_data
    
    # Visualize metaverse performance
    visualize_metaverse_performance(metaverse, simulation_data)
except Exception as e:
    print(f'Error in cell: {e}')
```
