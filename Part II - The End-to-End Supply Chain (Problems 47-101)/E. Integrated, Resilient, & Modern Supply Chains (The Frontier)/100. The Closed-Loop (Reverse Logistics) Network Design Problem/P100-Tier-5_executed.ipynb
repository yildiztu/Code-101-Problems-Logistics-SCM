# 100. The Closed-Loop Network Design Problem
## Tier 5: The Integrated Digital Twin

### Key Assumptions
- Real-time synchronization between physical and virtual systems
- IoT sensors provide continuous data streams from physical operations
- Digital twin can simulate scenarios and predict outcomes
- Predictive analytics enable proactive decision making
- System-of-systems integration across multiple logistics functions
- Continuous learning and adaptation from operational data

### Approach (Step-by-Step)
1. **Digital Twin Architecture**: Design comprehensive virtual representation
2. **IoT Sensor Integration**: Implement real-time data collection systems
3. **Synchronization Engine**: Ensure physical-virtual state alignment
4. **Predictive Analytics**: Develop forecasting and optimization models
5. **Scenario Simulation**: Enable what-if analysis and planning
6. **Continuous Learning**: Implement adaptive improvement mechanisms

### What to Look for in Results
- Real-time synchronization metrics and performance
- Predictive accuracy and forecasting capabilities
- Scenario analysis and optimization outcomes
- System integration and coordination effectiveness
- Learning curves and adaptation progress
- Digital twin fidelity and reliability measures

### Concrete Example
We'll implement:
- Digital twin with 50+ IoT sensors monitoring network operations
- Real-time synchronization with <1 second latency
- Predictive analytics for demand forecasting and optimization
- Scenario simulation for network reconfiguration
- Continuous learning from operational data

```python
from typing import Tuple, List, Dict, Set
from itertools import product

# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass, field
from typing import List, Tuple, Optional, Dict, Any
import random
from collections import defaultdict, deque
import warnings
from enum import Enum
import time
warnings.filterwarnings('ignore')

# Set style for better visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)
```

```python
class SensorType(Enum):
    """Types of IoT sensors in the digital twin"""
    TEMPERATURE = "temperature"
    HUMIDITY = "humidity"
    MOTION = "motion"
    INVENTORY = "inventory"
    TRAFFIC = "traffic"
    ENERGY = "energy"
    QUALITY = "quality"

@dataclass
class IoTSensor:
    """IoT sensor for real-time data collection"""
    sensor_id: str
    sensor_type: SensorType
    location: Tuple[float, float]
    value: float = 0.0
    last_update: float = 0.0
    accuracy: float = 0.95
    status: str = "active"  # 'active', 'maintenance', 'offline'
    
    def update_value(self, new_value: float, timestamp: float):
        """Update sensor value with timestamp"""
        self.value = new_value
        self.last_update = timestamp
        
        # Add realistic noise
        noise = np.random.normal(0, 0.02)
        self.value += noise

@dataclass
class DigitalEntity:
    """Digital representation of a physical entity"""
    entity_id: str
    entity_type: str  # 'facility', 'vehicle', 'product', 'customer'
    location: Tuple[float, float]
    physical_state: Dict[str, Any] = field(default_factory=dict)
    virtual_state: Dict[str, Any] = field(default_factory=dict)
    last_sync: float = 0.0
    sync_quality: float = 1.0
    
    def synchronize(self, timestamp: float) -> float:
        """Synchronize physical and virtual states"""
        
        # Calculate synchronization quality based on time since last sync
        time_diff = timestamp - self.last_sync
        if time_diff == 0:
            self.sync_quality = 1.0
        else:
            # Quality decreases with time, but improves with frequent syncs
            self.sync_quality = max(0.5, 1.0 - time_diff / 3600)  # 1 hour decay
        
        self.last_sync = timestamp
        
        # Simulate state differences
        for key in self.physical_state:
            if key not in self.virtual_state:
                self.virtual_state[key] = self.physical_state[key]
            else:
                # Add some divergence
                divergence = np.random.normal(0, 0.05) * self.sync_quality
                self.virtual_state[key] = self.physical_state[key] * (1 + divergence)
        
        return self.sync_quality

@dataclass
class PredictiveModel:
    """Predictive analytics model for forecasting"""
    model_id: str
    model_type: str  # 'demand_forecast', 'route_optimization', 'maintenance_prediction'
    accuracy: float = 0.8
    training_data: List[Dict[str, Any]] = field(default_factory=list)
    predictions: List[Dict[str, Any]] = field(default_factory=list)
    
    def train(self, data: List[Dict[str, Any]]):
        """Train the predictive model"""
        self.training_data.extend(data)
        
        # Simulate model training - accuracy improves with more data
        if len(self.training_data) > 100:
            self.accuracy = min(0.95, self.accuracy + 0.01)
    
    def predict(self, input_data: Dict[str, Any]) -> Dict[str, Any]:
        """Make prediction based on input data"""
        
        # Simulate prediction with some uncertainty
        if self.model_type == 'demand_forecast':
            base_demand = input_data.get('current_demand', 100)
            predicted_demand = base_demand * np.random.uniform(0.8, 1.2)
            confidence = self.accuracy * np.random.uniform(0.8, 1.0)
            
            return {
                'predicted_demand': predicted_demand,
                'confidence': confidence,
                'timestamp': time.time()
            }
        elif self.model_type == 'route_optimization':
            current_cost = input_data.get('current_cost', 1000)
            optimized_cost = current_cost * np.random.uniform(0.7, 0.95)
            
            return {
                'optimized_cost': optimized_cost,
                'savings': current_cost - optimized_cost,
                'timestamp': time.time()
            }
        else:
            return {
                'prediction': np.random.uniform(0, 1),
                'confidence': self.accuracy,
                'timestamp': time.time()
            }

@dataclass
class DigitalTwin:
    """Integrated Digital Twin system"""
    twin_id: str
    sensors: List[IoTSensor] = field(default_factory=list)
    entities: List[DigitalEntity] = field(default_factory=list)
    predictive_models: List[PredictiveModel] = field(default_factory=list)
    synchronization_status: float = 1.0
    prediction_accuracy: float = 0.8
    operational_metrics: Dict[str, float] = field(default_factory=dict)
    
    def add_sensor(self, sensor: IoTSensor):
        """Add an IoT sensor to the digital twin"""
        self.sensors.append(sensor)
    
    def add_entity(self, entity: DigitalEntity):
        """Add a digital entity to the twin"""
        self.entities.append(entity)
    
    def add_predictive_model(self, model: PredictiveModel):
        """Add a predictive model to the twin"""
        self.predictive_models.append(model)
    
    def update_sensors(self, timestamp: float):
        """Update all sensor readings"""
        
        for sensor in self.sensors:
            if sensor.status == 'active':
                # Generate realistic sensor values
                if sensor.sensor_type == SensorType.TEMPERATURE:
                    base_value = 20 + 5 * np.sin(timestamp / 3600 * 2 * np.pi)
                    new_value = base_value + np.random.normal(0, 2)
                elif sensor.sensor_type == SensorType.HUMIDITY:
                    base_value = 50 + 10 * np.sin(timestamp / 3600 * 2 * np.pi + np.pi/4)
                    new_value = base_value + np.random.normal(0, 5)
                elif sensor.sensor_type == SensorType.MOTION:
                    new_value = np.random.uniform(0, 100)  # Motion activity level
                elif sensor.sensor_type == SensorType.INVENTORY:
                    new_value = np.random.uniform(0, 1000)  # Inventory level
                elif sensor.sensor_type == SensorType.TRAFFIC:
                    new_value = np.random.uniform(0, 50)  # Traffic flow
                elif sensor.sensor_type == SensorType.ENERGY:
                    new_value = np.random.uniform(100, 500)  # Energy consumption
                elif sensor.sensor_type == SensorType.QUALITY:
                    new_value = np.random.uniform(0.8, 1.0)  # Quality score
                else:
                    new_value = np.random.uniform(0, 100)
                
                sensor.update_value(new_value, timestamp)
    
    def synchronize_entities(self, timestamp: float) -> float:
        """Synchronize all digital entities"""
        
        sync_qualities = []
        for entity in self.entities:
            sync_quality = entity.synchronize(timestamp)
            sync_qualities.append(sync_quality)
        
        if sync_qualities:
            self.synchronization_status = np.mean(sync_qualities)
        else:
            self.synchronization_status = 1.0
        
        return self.synchronization_status
    
    def run_predictions(self) -> List[Dict[str, Any]]:
        """Run all predictive models"""
        
        predictions = []
        
        for model in self.predictive_models:
            # Prepare input data from sensors
            input_data = {}
            
            for sensor in self.sensors:
                if sensor.status == 'active':
                    input_data[sensor.sensor_type.value] = sensor.value
            
            # Make prediction
            prediction = model.predict(input_data)
            prediction['model_id'] = model.model_id
            prediction['model_type'] = model.model_type
            predictions.append(prediction)
        
        return predictions
    
    def simulate_scenario(self, scenario_config: Dict[str, Any]) -> Dict[str, Any]:
        """Simulate a what-if scenario"""
        
        scenario_type = scenario_config.get('type', 'demand_increase')
        scenario_magnitude = scenario_config.get('magnitude', 1.2)
        
        # Simulate scenario impact
        if scenario_type == 'demand_increase':
            impact_on_cost = scenario_magnitude * 1.15
            impact_on_service = 1.0 / scenario_magnitude
        elif scenario_type == 'facility_disruption':
            impact_on_cost = 1.5
            impact_on_service = 0.7
        elif scenario_type == 'route_optimization':
            impact_on_cost = 0.85
            impact_on_service = 1.05
        else:
            impact_on_cost = 1.0
            impact_on_service = 1.0
        
        # Calculate scenario outcomes
        base_cost = self.operational_metrics.get('total_cost', 10000)
        base_service = self.operational_metrics.get('service_level', 0.95)
        
        scenario_outcome = {
            'scenario_type': scenario_type,
            'scenario_magnitude': scenario_magnitude,
            'predicted_cost': base_cost * impact_on_cost,
            'predicted_service_level': base_service * impact_on_service,
            'cost_impact': (impact_on_cost - 1.0) * 100,  # percentage
            'service_impact': (impact_on_service - 1.0) * 100,  # percentage
            'recommendation': self._generate_scenario_recommendation(scenario_type, impact_on_cost, impact_on_service)
        }
        
        return scenario_outcome
    
    def _generate_scenario_recommendation(self, scenario_type: str, cost_impact: float, service_impact: float) -> str:
        """Generate recommendation based on scenario analysis"""
        
        if cost_impact > 1.2:
            return "High cost impact - consider capacity expansion or cost optimization"
        elif service_impact < 0.8:
            return "Service level degradation - implement service recovery measures"
        elif cost_impact < 0.9 and service_impact > 1.0:
            return "Positive scenario - consider scaling successful strategies"
        else:
            return "Monitor situation and prepare contingency plans"
    
    def update_operational_metrics(self):
        """Update operational metrics based on current state"""
        
        # Calculate metrics from sensors
        active_sensors = [s for s in self.sensors if s.status == 'active']
        
        if active_sensors:
            avg_temperature = np.mean([s.value for s in active_sensors if s.sensor_type == SensorType.TEMPERATURE])
            avg_humidity = np.mean([s.value for s in active_sensors if s.sensor_type == SensorType.HUMIDITY])
            total_energy = sum([s.value for s in active_sensors if s.sensor_type == SensorType.ENERGY])
            avg_quality = np.mean([s.value for s in active_sensors if s.sensor_type == SensorType.QUALITY])
        else:
            avg_temperature = 20.0
            avg_humidity = 50.0
            total_energy = 0.0
            avg_quality = 0.9
        
        self.operational_metrics = {
            'total_cost': np.random.uniform(8000, 12000),
            'service_level': avg_quality,
            'energy_consumption': total_energy,
            'temperature': avg_temperature,
            'humidity': avg_humidity,
            'sensor_coverage': len(active_sensors) / len(self.sensors) if self.sensors else 0,
            'entity_sync_quality': self.synchronization_status,
            'prediction_accuracy': self.prediction_accuracy
        }
    
    def generate_twin_report(self) -> Dict[str, Any]:
        """Generate comprehensive digital twin report"""
        
        report = {
            'report_timestamp': time.time(),
            'twin_id': self.twin_id,
            'system_overview': {
                'total_sensors': len(self.sensors),
                'active_sensors': len([s for s in self.sensors if s.status == 'active']),
                'total_entities': len(self.entities),
                'total_models': len(self.predictive_models),
                'synchronization_status': self.synchronization_status,
                'prediction_accuracy': self.prediction_accuracy
            },
            'sensor_analysis': {},
            'entity_analysis': {},
            'model_analysis': {},
            'operational_metrics': self.operational_metrics,
            'recommendations': []
        }
        
        # Sensor analysis
        sensor_types = defaultdict(int)
        for sensor in self.sensors:
            sensor_types[sensor.sensor_type.value] += 1
        
        report['sensor_analysis'] = {
            'sensor_types': dict(sensor_types),
            'average_accuracy': np.mean([s.accuracy for s in self.sensors]) if self.sensors else 0,
            'coverage_area': len(self.sensors) * 10,  # Simplified coverage
        }
        
        # Entity analysis
        entity_types = defaultdict(int)
        for entity in self.entities:
            entity_types[entity.entity_type] += 1
        
        report['entity_analysis'] = {
            'entity_types': dict(entity_types),
            'average_sync_quality': np.mean([e.sync_quality for e in self.entities]) if self.entities else 0,
            'total_entities': len(self.entities)
        }
        
        # Model analysis
        model_types = defaultdict(int)
        for model in self.predictive_models:
            model_types[model.model_type] += 1
        
        report['model_analysis'] = {
            'model_types': dict(model_types),
            'average_accuracy': np.mean([m.accuracy for m in self.predictive_models]) if self.predictive_models else 0,
            'total_predictions': len([p for m in self.predictive_models for p in m.predictions])
        }
        
        # Generate recommendations
        if self.synchronization_status < 0.8:
            report['recommendations'].append('Improve synchronization frequency and quality')
        
        if len([s for s in self.sensors if s.status == 'active']) < len(self.sensors) * 0.8:
            report['recommendations'].append('Maintain sensor health and replace faulty units')
        
        if self.prediction_accuracy < 0.8:
            report['recommendations'].append('Retrain predictive models with more data')
        
        if self.operational_metrics.get('service_level', 0) < 0.9:
            report['recommendations'].append('Implement service improvement measures')
        
        return report
```

```python
# Initialize Digital Twin System
print("Initializing Integrated Digital Twin System...")
print("=" * 50)

digital_twin = DigitalTwin(twin_id="closed_loop_twin_001")

# Create IoT sensors
sensor_types = [SensorType.TEMPERATURE, SensorType.HUMIDITY, SensorType.MOTION, 
                SensorType.INVENTORY, SensorType.TRAFFIC, SensorType.ENERGY, SensorType.QUALITY]
num_sensors_per_type = 8  # 8 sensors per type = 56 total sensors

print(f"Creating {len(sensor_types) * num_sensors_per_type} IoT sensors...")

for sensor_type in sensor_types:
    for i in range(num_sensors_per_type):
        sensor = IoTSensor(
            sensor_id=f"{sensor_type.value}_{i}",
            sensor_type=sensor_type,
            location=(np.random.uniform(0, 100), np.random.uniform(0, 100)),
            accuracy=np.random.uniform(0.9, 0.99)
        )
        digital_twin.add_sensor(sensor)

print(f"Sensors Created: {len(digital_twin.sensors)}")
for sensor_type in sensor_types:
    count = len([s for s in digital_twin.sensors if s.sensor_type == sensor_type])
    print(f"  {sensor_type.value.title()}: {count} sensors")

# Create digital entities
entity_configurations = [
    ("facility_1", "facility", (20, 30)),
    ("facility_2", "facility", (60, 70)),
    ("facility_3", "facility", (40, 50)),
    ("vehicle_1", "vehicle", (25, 35)),
    ("vehicle_2", "vehicle", (45, 55)),
    ("vehicle_3", "vehicle", (65, 75)),
    ("product_a", "product", (30, 40)),
    ("product_b", "product", (50, 60)),
    ("customer_1", "customer", (15, 25)),
    ("customer_2", "customer", (55, 65)),
]

print(f"\nCreating Digital Entities...")

for entity_id, entity_type, location in entity_configurations:
    entity = DigitalEntity(
        entity_id=entity_id,
        entity_type=entity_type,
        location=location,
        physical_state={
            'status': 'active',
            'capacity': np.random.uniform(100, 1000),
            'utilization': np.random.uniform(0.3, 0.9),
            'efficiency': np.random.uniform(0.7, 0.95)
        }
    )
    digital_twin.add_entity(entity)

print(f"Digital Entities Created: {len(digital_twin.entities)}")
for entity_type in ['facility', 'vehicle', 'product', 'customer']:
    count = len([e for e in digital_twin.entities if e.entity_type == entity_type])
    print(f"  {entity_type.title()}: {count} entities")

# Create predictive models
model_configurations = [
    ("demand_forecast_001", "demand_forecast"),
    ("route_optimization_001", "route_optimization"),
    ("maintenance_prediction_001", "maintenance_prediction"),
]

print(f"\nCreating Predictive Models...")

for model_id, model_type in model_configurations:
    model = PredictiveModel(
        model_id=model_id,
        model_type=model_type,
        accuracy=np.random.uniform(0.75, 0.9)
    )
    digital_twin.add_predictive_model(model)

print(f"Predictive Models Created: {len(digital_twin.predictive_models)}")
for model in digital_twin.predictive_models:
    print(f"  {model.model_id}: {model.model_type.title()} (accuracy: {model.accuracy:.3f})")

# Initialize operational metrics
digital_twin.update_operational_metrics()

print(f"\nDigital Twin System Ready:")
print(f"  Twin ID: {digital_twin.twin_id}")
print(f"  Total Sensors: {len(digital_twin.sensors)}")
print(f"  Total Entities: {len(digital_twin.entities)}")
print(f"  Total Models: {len(digital_twin.predictive_models)}")
print(f"  Synchronization Status: {digital_twin.synchronization_status:.3f}")
print(f"  Prediction Accuracy: {digital_twin.prediction_accuracy:.3f}")
```

```python
try:
    def run_digital_twin_simulation(digital_twin: DigitalTwin, duration_minutes: int = 5):
        """Run digital twin simulation with real-time updates"""
        
        print(f"Running Digital Twin Simulation for {duration_minutes} minutes...")
        print("=" * 60)
        
        simulation_data = {
            'timestamps': [],
            'synchronization_status': [],
            'sensor_readings': [],
            'predictions': [],
            'operational_metrics': [],
            'scenario_simulations': []
        }
        
        start_time = time.time()
        
        for minute in range(duration_minutes):
            for second in range(60):
                current_time = time.time()
                
                # Update sensors
                digital_twin.update_sensors(current_time)
                
                # Synchronize entities
                sync_status = digital_twin.synchronize_entities(current_time)
                
                # Run predictions (every 10 seconds)
                if second % 10 == 0:
                    predictions = digital_twin.run_predictions()
                    
                    # Train models with new data
                    for model in digital_twin.predictive_models:
                        sensor_data = [{s.sensor_type.value: s.value for s in digital_twin.sensors if s.status == 'active'}]
                        model.train(sensor_data)
                else:
                    predictions = []
                
                # Update operational metrics
                digital_twin.update_operational_metrics()
                
                # Run scenario simulation (every 30 seconds)
                if second % 30 == 0:
                    scenarios = [
                        {'type': 'demand_increase', 'magnitude': 1.2},
                        {'type': 'facility_disruption', 'magnitude': 1.0},
                        {'type': 'route_optimization', 'magnitude': 1.0}
                    ]
                    scenario = np.random.choice(scenarios)
                    scenario_result = digital_twin.simulate_scenario(scenario)
                else:
                    scenario_result = None
                
                # Collect data
                simulation_data['timestamps'].append(current_time)
                simulation_data['synchronization_status'].append(sync_status)
                simulation_data['sensor_readings'].append({s.sensor_id: s.value for s in digital_twin.sensors if s.status == 'active'})
                simulation_data['predictions'].append(predictions)
                simulation_data['operational_metrics'].append(digital_twin.operational_metrics.copy())
                simulation_data['scenario_simulations'].append(scenario_result)
                
                # Print status every minute
                if second == 0:
                    print(f"Minute {minute + 1}: Sync={sync_status:.3f}, Sensors={len([s for s in digital_twin.sensors if s.status == 'active'])}, Entities={len(digital_twin.entities)}")
                    
                    # Show sample predictions
                    if predictions:
                        print(f"  Sample Predictions: {len(predictions)} total")
                        for prediction in predictions[:2]:
                            if prediction['model_type'] == 'demand_forecast':
                                print(f"    {prediction['model_id']}: Demand={prediction['predicted_demand']:.1f} (confidence: {prediction['confidence']:.2f})")
                            elif prediction['model_type'] == 'route_optimization':
                                print(f"    {prediction['model_id']}: Cost=${prediction['optimized_cost']:.0f} (savings: ${prediction['savings']:.0f})")
                    
                    # Show scenario if available
                    if scenario_result:
                        print(f"  Scenario: {scenario_result['scenario_type'].replace('_', ' ').title()}")
                        print(f"    Cost Impact: {scenario_result['cost_impact']:+.1f}%, Service Impact: {scenario_result['service_impact']:+.1f}%")
                
                time.sleep(0.01)  # Small delay for realistic simulation
        
        end_time = time.time()
        
        print(f"\nSimulation Summary:")
        print(f"Duration: {end_time - start_time:.2f} seconds")
        print(f"Total Updates: {len(simulation_data['timestamps'])}")
        print(f"Average Synchronization: {np.mean(simulation_data['synchronization_status']):.3f}")
        print(f"Total Predictions: {sum(len(p) for p in simulation_data['predictions'])}")
        print(f"Total Scenarios: {len([s for s in simulation_data['scenario_simulations'] if s is not None])}")
        
        return simulation_data
    
    # Run the simulation
    simulation_data = run_digital_twin_simulation(digital_twin, duration_minutes=3)
    
    # Generate final report
    final_report = digital_twin.generate_twin_report()
    
    print(f"\nFinal Digital Twin Report:")
    print("=" * 30)
    
    overview = final_report['system_overview']
    print(f"System Overview:")
    print(f"  Total Sensors: {overview['total_sensors']}")
    print(f"  Active Sensors: {overview['active_sensors']} ({overview['active_sensors']/overview['total_sensors']:.1%})")
    print(f"  Total Entities: {overview['total_entities']}")
    print(f"  Total Models: {overview['total_models']}")
    print(f"  Synchronization Status: {overview['synchronization_status']:.3f}")
    print(f"  Prediction Accuracy: {overview['prediction_accuracy']:.3f}")
    
    print(f"\nOperational Metrics:")
    for metric, value in final_report['operational_metrics'].items():
        if isinstance(value, float):
            print(f"  {metric.replace('_', ' ').title()}: {value:.3f}")
        else:
            print(f"  {metric.replace('_', ' ').title()}: {value}")
    
    print(f"\nRecommendations:")
    for i, recommendation in enumerate(final_report['recommendations'], 1):
        print(f"  {i}. {recommendation}")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    def visualize_digital_twin_performance(digital_twin: DigitalTwin, 
                                         simulation_data: Dict[str, List]):
        """Visualize digital twin performance and analytics"""
        
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
        
        # Plot 1: Synchronization status over time
        ax1.set_title('Digital Twin Synchronization Status', fontsize=14, fontweight='bold')
        
        time_minutes = [(t - simulation_data['timestamps'][0]) / 60 for t in simulation_data['timestamps']]
        
        ax1.plot(time_minutes, simulation_data['synchronization_status'], 'b-', linewidth=2)
        ax1.axhline(y=0.8, color='r', linestyle='--', alpha=0.7, label='Good Sync Threshold')
        ax1.axhline(y=0.9, color='g', linestyle='--', alpha=0.7, label='Excellent Sync Threshold')
        ax1.set_xlabel('Time (minutes)')
        ax1.set_ylabel('Synchronization Status (0-1)')
        ax1.set_ylim(0, 1)
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # Plot 2: Sensor readings heatmap
        ax2.set_title('IoT Sensor Readings Heatmap', fontsize=14, fontweight='bold')
        
        # Create sensor data matrix
        active_sensors = [s for s in digital_twin.sensors if s.status == 'active']
        if active_sensors:
            sensor_matrix = []
            sensor_labels = []
            
            for sensor_type in SensorType:
                type_sensors = [s for s in active_sensors if s.sensor_type == sensor_type]
                if type_sensors:
                    # Average readings for this sensor type
                    readings = []
                    for reading_data in simulation_data['sensor_readings']:
                        type_readings = [reading_data[s.sensor_id] for s in type_sensors if s.sensor_id in reading_data]
                        if type_readings:
                            readings.append(np.mean(type_readings))
                        else:
                            readings.append(0)
                    sensor_matrix.append(readings)
                    sensor_labels.append(sensor_type.value.replace('_', ' ').title())
            
            if sensor_matrix:
                # Sample data for heatmap (every 10th reading)
                sample_indices = range(0, len(sensor_matrix[0]), max(1, len(sensor_matrix[0]) // 50))
                heatmap_data = [row[i] for i in sample_indices for row in sensor_matrix]
                heatmap_data = np.array(heatmap_data).reshape(len(sensor_matrix), len(sample_indices))
                
                im = ax2.imshow(heatmap_data, cmap='viridis', aspect='auto')
                ax2.set_yticks(range(len(sensor_labels)))
                ax2.set_yticklabels(sensor_labels)
                ax2.set_xlabel('Time (sampled)')
                
                # Add colorbar
                cbar = plt.colorbar(im, ax=ax2)
                cbar.set_label('Sensor Value')
        
        # Plot 3: Prediction accuracy over time
        ax3.set_title('Predictive Model Performance', fontsize=14, fontweight='bold')
        
        # Extract prediction accuracies
        prediction_times = []
        demand_predictions = []
        route_predictions = []
        
        for i, predictions in enumerate(simulation_data['predictions']):
            if predictions:
                prediction_times.append(time_minutes[i])
                for pred in predictions:
                    if pred['model_type'] == 'demand_forecast':
                        demand_predictions.append(pred['confidence'])
                    elif pred['model_type'] == 'route_optimization':
                        route_predictions.append(pred.get('confidence', 0.8))
        
        if prediction_times:
            if demand_predictions:
                ax3.plot(prediction_times[:len(demand_predictions)], demand_predictions, 'g-', label='Demand Forecast', linewidth=2, marker='o', markersize=3)
            if route_predictions:
                ax3.plot(prediction_times[:len(route_predictions)], route_predictions, 'orange', label='Route Optimization', linewidth=2, marker='s', markersize=3)
            
            ax3.set_xlabel('Time (minutes)')
            ax3.set_ylabel('Prediction Confidence (0-1)')
            ax3.set_ylim(0, 1)
            ax3.legend()
            ax3.grid(True, alpha=0.3)
        
        # Plot 4: Operational metrics evolution
        ax4.set_title('Operational Metrics Evolution', fontsize=14, fontweight='bold')
        
        # Extract key metrics
        service_levels = [metrics.get('service_level', 0) for metrics in simulation_data['operational_metrics']]
        energy_consumption = [metrics.get('energy_consumption', 0) / 1000 for metrics in simulation_data['operational_metrics']]  # Scale down
        
        ax4_twin = ax4.twinx()
        
        line1 = ax4.plot(time_minutes, service_levels, 'b-', label='Service Level', linewidth=2)
        ax4.set_xlabel('Time (minutes)')
        ax4.set_ylabel('Service Level (0-1)', color='b')
        ax4.tick_params(axis='y', labelcolor='b')
        ax4.set_ylim(0, 1)
        
        line2 = ax4_twin.plot(time_minutes, energy_consumption, 'r-', label='Energy Consumption', linewidth=2)
        ax4_twin.set_ylabel('Energy Consumption (kW)', color='r')
        ax4_twin.tick_params(axis='y', labelcolor='r')
        
        ax4.legend([line1, line2], ['Service Level', 'Energy Consumption'], loc='upper left')
        ax4.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
        
        # Additional analysis
        print("\nDigital Twin Performance Analysis:")
        print("=" * 40)
        
        # Calculate system health score
        avg_sync = np.mean(simulation_data['synchronization_status'])
        avg_service = np.mean(service_levels)
        total_predictions = sum(len(p) for p in simulation_data['predictions'])
        
        system_health = (avg_sync + avg_service) / 2
        
        print(f"System Health Score: {system_health:.3f}")
        print(f"  Synchronization: {avg_sync:.3f}")
        print(f"  Service Level: {avg_service:.3f}")
        
        # Sensor performance analysis
        active_sensors = len([s for s in digital_twin.sensors if s.status == 'active'])
        sensor_coverage = active_sensors / len(digital_twin.sensors) if digital_twin.sensors else 0
        
        print(f"\nSensor Performance:")
        print(f"  Active Sensors: {active_sensors}/{len(digital_twin.sensors)} ({sensor_coverage:.1%})")
        print(f"  Sensor Coverage: {sensor_coverage:.3f}")
        
        # Prediction analysis
        print(f"\nPrediction Analysis:")
        print(f"  Total Predictions: {total_predictions}")
        print(f"  Predictions per Minute: {total_predictions / len(simulation_data['timestamps']) * 60:.1f}")
        
        if demand_predictions:
            avg_demand_confidence = np.mean(demand_predictions)
            print(f"  Demand Forecast Confidence: {avg_demand_confidence:.3f}")
        
        if route_predictions:
            avg_route_confidence = np.mean(route_predictions)
            print(f"  Route Optimization Confidence: {avg_route_confidence:.3f}")
        
        # Scenario analysis
        scenarios = [s for s in simulation_data['scenario_simulations'] if s is not None]
        if scenarios:
            scenario_types = [s['scenario_type'] for s in scenarios]
            type_counts = defaultdict(int)
            for scenario_type in scenario_types:
                type_counts[scenario_type] += 1
            
            print(f"\nScenario Analysis:")
            print(f"  Total Scenarios: {len(scenarios)}")
            for scenario_type, count in type_counts.items():
                print(f"    {scenario_type.replace('_', ' ').title()}: {count}")
        
        # Assessment
        print(f"\nOverall Assessment:")
        if system_health > 0.85:
            print(f"Status: EXCELLENT - Digital twin operating at peak performance")
        elif system_health > 0.75:
            print(f"Status: GOOD - Digital twin performing well with minor issues")
        elif system_health > 0.65:
            print(f"Status: ACCEPTABLE - Digital twin functional but needs optimization")
        else:
            print(f"Status: NEEDS IMPROVEMENT - Digital twin requires significant optimization")
        
        return simulation_data
    
    # Visualize digital twin performance
    visualize_digital_twin_performance(digital_twin, simulation_data)
except Exception as e:
    print(f'Error in cell: {e}')
```

- **Scenario Simulation**: Test what-if scenarios without physical risks
- **System Integration**: Holistic view across multiple logistics functions
- **Continuous Learning**: Adaptive improvement from operational data
- **IoT Integration**: Real-time data collection from physical operations
