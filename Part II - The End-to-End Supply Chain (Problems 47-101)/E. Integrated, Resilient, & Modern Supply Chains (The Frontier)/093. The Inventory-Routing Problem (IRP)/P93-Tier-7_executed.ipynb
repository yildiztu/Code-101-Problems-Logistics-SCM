# 93. The Inventory-Routing Problem (IRP)
## Tier 7: The Human-AI Symbiotic Partnership (The Centaur Model)

### Key assumptions
- Human-AI collaborative decision-making
- Explainable AI with transparent reasoning
- Trust-based relationship development
- Human-in-the-loop for critical decisions
- Continuous learning from human expertise

### Approach (step-by-step)
The Human-AI Symbiotic Partnership creates collaborative intelligence:

1. **Explainable AI System**:
   - Transparent decision reasoning
   - Interpretable model explanations
   - Confidence quantification
   - Alternative solution generation

2. **Human Expert Interface**:
   - Intuitive visualization dashboards
   - Interactive decision support
   - Expert knowledge capture
   - Feedback integration mechanisms

3. **Trust Development Framework**:
   - Performance tracking and validation
   - Reliability assessment
   - Confidence calibration
   - Relationship metrics monitoring

4. **Collaborative Decision Process**:
   - AI generates initial recommendations
   - Human reviews and provides feedback
   - System learns from human corrections
   - Joint decision finalization

5. **Learning and Adaptation**:
   - Expert pattern recognition
   - Preference learning
   - Decision quality improvement
   - Relationship strengthening

### What to look for in the results
- Human-AI agreement rates
- Decision quality improvement over time
- Trust development metrics
- Learning curve progression
- Expert knowledge capture effectiveness

### Concrete example (from the source)
Advanced instance with:
- 1 depot, 15 customers
- 8-day planning horizon
- 4 vehicles with collaborative routing
- Human expert in the decision loop
- Explainable AI recommendations
- Trust-based decision framework

### Why this Tier exists vs Tiers 1-6
Human-AI Partnership addresses limitations of pure automation:
- **Human Expertise**: Domain knowledge vs. algorithmic optimization
- **Contextual Understanding**: Real-world context vs. mathematical models
- **Trust and Acceptance**: Human confidence vs. black-box decisions
- **Continuous Learning**: Expert feedback vs. static algorithms

### Pros / Cons vs earlier Tiers
**Pros:**
- Superior decision quality through human expertise
- Higher trust and acceptance in practice
- Continuous learning and improvement
- Contextual awareness and adaptability
- Explainable and transparent decisions

**Cons:**
- Requires human time and expertise
- Slower decision-making process
- Dependence on human availability
- Potential for human bias
- Implementation complexity

### When to use this Tier
- High-stakes inventory routing decisions
- Complex, context-dependent environments
- Organizations with domain experts
- When trust and explainability are critical
- Continuous improvement environments

```python
from typing import Tuple, List, Dict, Set

# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass
from typing import Dict, List, Tuple, Optional, Any
import itertools
import random
import time
from collections import defaultdict
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
@dataclass
class AIRecommendation:
    """AI-generated recommendation with explanations"""
    recommendation_type: str  # 'route', 'inventory', 'timing'
    action: str
    confidence: float  # 0-1
    reasoning: List[str]  # Step-by-step reasoning
    alternatives: List[Dict[str, Any]]  # Alternative options
    expected_impact: Dict[str, float]  # Expected outcomes
    risk_assessment: Dict[str, float]  # Risk factors

@dataclass
class HumanFeedback:
    """Human expert feedback on AI recommendations"""
    recommendation_id: str
    accepted: bool
    modifications: Dict[str, Any]
    reasoning: str
    confidence_in_decision: float
    expert_notes: str

@dataclass
class TrustMetrics:
    """Trust and relationship metrics"""
    overall_trust: float  # 0-1
    prediction_accuracy: float  # 0-1
    reliability_score: float  # 0-1
    agreement_rate: float  # 0-1
    learning_progress: float  # 0-1
    relationship_strength: float  # 0-1

class ExplainableAI:
    """Explainable AI system for IRP decisions"""
    
    def __init__(self, depot, customers, vehicles, num_periods):
        self.depot = depot
        self.customers = customers
        self.vehicles = vehicles
        self.num_periods = num_periods
        
        # Learning components
        self.expert_patterns = []
        self.decision_history = []
        self.performance_history = []
        
        # Current state tracking
        self.current_inventory = {}
        self.current_routes = {}
        self.current_costs = {}
        
        # Initialize state
        self._initialize_state()
    
    def _initialize_state(self):
        """Initialize current system state"""
        # Customer inventories
        for customer in self.customers:
            self.current_inventory[customer.id] = customer.initial_inventory
        
        # Vehicle routes
        for vehicle in self.vehicles:
            self.current_routes[vehicle.id] = []
            self.current_costs[vehicle.id] = 0
    
    def generate_route_recommendation(self, period: int, vehicle_id: int) -> AIRecommendation:
        """Generate route recommendation with explanation"""
        vehicle = next(v for v in self.vehicles if v.id == vehicle_id)
        
        # Analyze current situation
        urgent_customers = self._identify_urgent_customers()
        nearby_customers = self._find_nearby_customers(vehicle_id)
        capacity_constraints = self._assess_capacity_constraints(vehicle_id)
        
        # Generate optimal route (simplified)
        recommended_route = self._calculate_optimal_route(vehicle_id, urgent_customers + nearby_customers)
        
        # Build reasoning
        reasoning = [
            f"Vehicle {vehicle_id} has capacity {vehicle.capacity} units",
            f"Identified {len(urgent_customers)} urgent customers requiring immediate delivery",
            f"Found {len(nearby_customers)} nearby customers for efficiency",
            f"Route minimizes total distance while serving urgent needs first",
            f"Expected travel distance: {self._calculate_route_distance(recommended_route):.1f} units"
        ]
        
        # Generate alternatives
        alternatives = [
            {
                'route': urgent_customers[:3],
                'reason': 'Focus only on most urgent customers',
                'impact': {'distance': '20% less', 'service_level': '10% lower'}
            },
            {
                'route': nearby_customers[:4],
                'reason': 'Focus on nearby customers for efficiency',
                'impact': {'distance': '30% less', 'service_level': '15% lower'}
            }
        ]
        
        # Calculate expected impact
        expected_impact = {
            'service_improvement': 25.0,
            'cost_reduction': 15.0,
            'efficiency_gain': 20.0,
            'risk_reduction': 10.0
        }
        
        # Risk assessment
        risk_assessment = {
            'demand_uncertainty': 0.3,
            'capacity_risk': 0.2,
            'travel_delay_risk': 0.15,
            'stockout_risk': 0.25
        }
        
        # Calculate confidence
        confidence = self._calculate_confidence(urgent_customers, nearby_customers, capacity_constraints)
        
        return AIRecommendation(
            recommendation_type='route',
            action=f"Route vehicle {vehicle_id} through customers: {recommended_route}",
            confidence=confidence,
            reasoning=reasoning,
            alternatives=alternatives,
            expected_impact=expected_impact,
            risk_assessment=risk_assessment
        )
    
    def generate_inventory_recommendation(self, period: int, customer_id: int) -> AIRecommendation:
        """Generate inventory management recommendation"""
        customer = next(c for c in self.customers if c.id == customer_id)
        current_inventory = self.current_inventory[customer_id]
        
        # Analyze inventory situation
        days_of_stock = current_inventory / customer.demand_mean
        urgency_level = self._calculate_urgency(customer_id)
        demand_forecast = self._forecast_demand(customer_id)
        
        # Generate recommendation
        if days_of_stock < 2:
            action = f"Urgent delivery needed: Order {customer.max_inventory - current_inventory:.1f} units"
            confidence = 0.9
        elif days_of_stock < 4:
            action = f"Schedule delivery: Order {customer.min_inventory * 2 - current_inventory:.1f} units"
            confidence = 0.8
        else:
            action = f"Maintain current inventory: No immediate delivery needed"
            confidence = 0.7
        
        # Build reasoning
        reasoning = [
            f"Customer {customer_id} has {current_inventory:.1f} units in stock",
            f"Current inventory covers {days_of_stock:.1f} days of demand",
            f"Average daily demand: {customer.demand_mean:.1f} units",
            f"Minimum inventory level: {customer.min_inventory:.1f} units",
            f"Demand forecast for next period: {demand_forecast:.1f} units",
            f"Urgency level: {urgency_level:.2f} (1=highest, 5=lowest)"
        ]
        
        # Generate alternatives
        alternatives = [
            {
                'action': 'Order larger quantity',
                'reason': 'Reduce future delivery frequency',
                'impact': {'holding_cost': '+15%', 'delivery_frequency': '-30%'}
            },
            {
                'action': 'Order smaller quantity',
                'reason': 'Minimize holding costs',
                'impact': {'holding_cost': '-20%', 'delivery_frequency': '+25%'}
            }
        ]
        
        # Calculate expected impact
        expected_impact = {
            'stockout_prevention': 30.0,
            'cost_optimization': 20.0,
            'service_improvement': 25.0,
            'efficiency_gain': 15.0
        }
        
        # Risk assessment
        risk_assessment = {
            'demand_spike_risk': 0.35,
            'holding_cost_risk': 0.2,
            'stockout_risk': 0.4,
            'forecast_error_risk': 0.25
        }
        
        return AIRecommendation(
            recommendation_type='inventory',
            action=action,
            confidence=confidence,
            reasoning=reasoning,
            alternatives=alternatives,
            expected_impact=expected_impact,
            risk_assessment=risk_assessment
        )
    
    def _identify_urgent_customers(self) -> List[int]:
        """Identify customers needing urgent delivery"""
        urgent = []
        for customer in self.customers:
            current = self.current_inventory[customer.id]
            days_of_stock = current / customer.demand_mean
            if days_of_stock < 2:  # Less than 2 days of stock
                urgent.append(customer.id)
        return urgent
    
    def _find_nearby_customers(self, vehicle_id: int) -> List[int]:
        """Find customers near vehicle's current location"""
        # Simplified: find customers within radius of depot
        nearby = []
        for customer in self.customers:
            distance = np.sqrt(customer.x**2 + customer.y**2)
            if distance < 30:  # Within 30 units
                nearby.append(customer.id)
        return nearby[:5]  # Return up to 5 nearby customers
    
    def _assess_capacity_constraints(self, vehicle_id: int) -> Dict[str, float]:
        """Assess vehicle capacity constraints"""
        vehicle = next(v for v in self.vehicles if v.id == vehicle_id)
        current_load = sum(self.current_inventory[c] for c in self.current_routes[vehicle_id])
        
        return {
            'available_capacity': vehicle.capacity - current_load,
            'utilization_rate': current_load / vehicle.capacity,
            'constraint_level': (current_load / vehicle.capacity) * 100
        }
    
    def _calculate_optimal_route(self, vehicle_id: int, customers: List[int]) -> List[int]:
        """Calculate optimal route for vehicle (simplified TSP)"""
        if not customers:
            return []
        
        # Simple nearest neighbor heuristic
        unvisited = customers.copy()
        route = []
        current_pos = (0, 0)  # Start at depot
        
        while unvisited:
            nearest_customer = None
            nearest_distance = float('inf')
            
            for customer_id in unvisited:
                customer = next(c for c in self.customers if c.id == customer_id)
                distance = np.sqrt((customer.x - current_pos[0])**2 + (customer.y - current_pos[1])**2)
                
                if distance < nearest_distance:
                    nearest_distance = distance
                    nearest_customer = customer_id
            
            if nearest_customer:
                route.append(nearest_customer)
                unvisited.remove(nearest_customer)
                customer = next(c for c in self.customers if c.id == nearest_customer)
                current_pos = (customer.x, customer.y)
        
        return route
    
    def _calculate_route_distance(self, route: List[int]) -> float:
        """Calculate total distance for a route"""
        if not route:
            return 0
        
        total_distance = 0
        current_pos = (0, 0)  # Start at depot
        
        for customer_id in route:
            customer = next(c for c in self.customers if c.id == customer_id)
            distance = np.sqrt((customer.x - current_pos[0])**2 + (customer.y - current_pos[1])**2)
            total_distance += distance
            current_pos = (customer.x, customer.y)
        
        # Return to depot
        total_distance += np.sqrt(current_pos[0]**2 + current_pos[1]**2)
        
        return total_distance
    
    def _calculate_confidence(self, urgent: List[int], nearby: List[int], constraints: Dict) -> float:
        """Calculate confidence in recommendation"""
        base_confidence = 0.7
        
        # Increase confidence with more urgent customers
        urgent_factor = min(1.0, len(urgent) / 3) * 0.2
        
        # Increase confidence with good capacity utilization
        utilization = constraints['utilization_rate']
        if 0.6 <= utilization <= 0.9:
            capacity_factor = 0.1
        else:
            capacity_factor = -0.1
        
        # Add learning from past performance
        learning_factor = 0.1 if len(self.performance_history) > 5 else 0
        
        confidence = base_confidence + urgent_factor + capacity_factor + learning_factor
        return max(0.5, min(0.95, confidence))
    
    def _calculate_urgency(self, customer_id: int) -> float:
        """Calculate urgency level for customer"""
        customer = next(c for c in self.customers if c.id == customer_id)
        current = self.current_inventory[customer_id]
        days_of_stock = current / customer.demand_mean
        
        if days_of_stock < 1:
            return 1.0  # Critical
        elif days_of_stock < 2:
            return 0.8  # High
        elif days_of_stock < 4:
            return 0.6  # Medium
        else:
            return 0.3  # Low
    
    def _forecast_demand(self, customer_id: int) -> float:
        """Forecast demand for next period"""
        customer = next(c for c in self.customers if c.id == customer_id)
        # Simple forecast with some randomness
        base_forecast = customer.demand_mean
        variation = np.random.normal(0, customer.demand_std * 0.3)
        return max(0, base_forecast + variation)
    
    def learn_from_feedback(self, feedback: HumanFeedback):
        """Learn from human expert feedback"""
        self.decision_history.append(feedback)
        
        # Extract patterns from expert decisions
        if feedback.accepted:
            self.expert_patterns.append({
                'context': 'accepted',
                'modifications': feedback.modifications,
                'reasoning': feedback.reasoning
            })
        
        # Update performance metrics
        self._update_performance_metrics()
    
    def _update_performance_metrics(self):
        """Update internal performance metrics"""
        if len(self.decision_history) > 0:
            recent_decisions = self.decision_history[-10:]  # Last 10 decisions
            acceptance_rate = sum(1 for d in recent_decisions if d.accepted) / len(recent_decisions)
            self.performance_history.append(acceptance_rate)

class HumanExpert:
    """Human expert domain knowledge and decision-making"""
    
    def __init__(self, expertise_level: float = 0.8):
        self.expertise_level = expertise_level  # 0-1
        self.decision_patterns = []
        self.preferences = {}
        self.risk_tolerance = 0.3  # 0-1
        
        # Expert knowledge base
        self.customer_priorities = {}
        self.route_preferences = {}
        self.inventory_strategies = {}
        
        # Initialize expert knowledge
        self._initialize_expert_knowledge()
    
    def _initialize_expert_knowledge(self):
        """Initialize domain expert knowledge"""
        # Customer priorities based on importance
        self.customer_priorities = {
            1: 0.9, 2: 0.8, 3: 0.7, 4: 0.6, 5: 0.8,
            6: 0.7, 7: 0.9, 8: 0.6, 9: 0.8, 10: 0.7,
            11: 0.8, 12: 0.7, 13: 0.6, 14: 0.9, 15: 0.8
        }
        
        # Route preferences (efficiency vs. service balance)
        self.route_preferences = {
            'efficiency_weight': 0.6,
            'service_weight': 0.4,
            'max_route_length': 4,
            'prefer_clustering': True
        }
        
        # Inventory strategies
        self.inventory_strategies = {
            'safety_stock_multiplier': 1.5,
            'reorder_point_multiplier': 1.2,
            'batch_size_preference': 'medium'
        }
    
    def review_recommendation(self, recommendation: AIRecommendation) -> HumanFeedback:
        """Review AI recommendation and provide feedback"""
        # Simulate expert decision-making
        expert_confidence = self._evaluate_recommendation(recommendation)
        
        # Make decision based on expertise and confidence
        if expert_confidence > 0.7 and recommendation.confidence > 0.7:
            # Accept recommendation
            accepted = True
            modifications = {}
            reasoning = "AI recommendation aligns with expert judgment and business priorities"
        else:
            # Modify or reject recommendation
            accepted = False
            modifications = self._generate_modifications(recommendation)
            reasoning = "Expert modification based on domain knowledge and business context"
        
        # Calculate confidence in final decision
        decision_confidence = min(1.0, expert_confidence + (recommendation.confidence * 0.3))
        
        # Generate expert notes
        expert_notes = self._generate_expert_notes(recommendation, accepted, modifications)
        
        return HumanFeedback(
            recommendation_id=f"REC_{int(time.time())}",
            accepted=accepted,
            modifications=modifications,
            reasoning=reasoning,
            confidence_in_decision=decision_confidence,
            expert_notes=expert_notes
        )
    
    def _evaluate_recommendation(self, recommendation: AIRecommendation) -> float:
        """Evaluate AI recommendation based on expert knowledge"""
        base_score = 0.5
        
        # Evaluate reasoning quality
        if len(recommendation.reasoning) >= 4:
            base_score += 0.1
        
        # Evaluate alternatives
        if len(recommendation.alternatives) >= 2:
            base_score += 0.1
        
        # Evaluate risk assessment
        if max(recommendation.risk_assessment.values()) < 0.5:
            base_score += 0.1
        
        # Add expertise-based evaluation
        expertise_bonus = self.expertise_level * 0.2
        
        return min(1.0, base_score + expertise_bonus)
    
    def _generate_modifications(self, recommendation: AIRecommendation) -> Dict[str, Any]:
        """Generate expert modifications to recommendation"""
        modifications = {}
        
        if recommendation.recommendation_type == 'route':
            # Modify route based on expert preferences
            modifications['route_adjustment'] = 'Prioritize high-value customers'
            modifications['efficiency_tradeoff'] = 'Accept 10% longer routes for better service'
        elif recommendation.recommendation_type == 'inventory':
            # Modify inventory based on expert strategies
            modifications['safety_stock'] = 'Increase safety stock by 20%'
            modifications['delivery_frequency'] = 'Reduce delivery frequency for efficiency'
        
        return modifications
    
    def _generate_expert_notes(self, recommendation: AIRecommendation, 
                              accepted: bool, modifications: Dict) -> str:
        """Generate detailed expert notes"""
        if accepted:
            notes = (f"Accepted AI {recommendation.recommendation_type} recommendation. "
                   f"Confidence: {recommendation.confidence:.2f}. "
                   f"Reasoning: {'; '.join(recommendation.reasoning[:2])}.")
        else:
            notes = (f"Modified AI {recommendation.recommendation_type} recommendation. "
                   f"Applied expert modifications: {list(modifications.keys())}. "
                   f"Reasoning: Expert domain knowledge takes precedence.")
        
        return notes

class HumanAIPartnership:
    """Human-AI Symbiotic Partnership system"""
    
    def __init__(self, depot, customers, vehicles, num_periods):
        self.depot = depot
        self.customers = customers
        self.vehicles = vehicles
        self.num_periods = num_periods
        
        # Initialize components
        self.ai_system = ExplainableAI(depot, customers, vehicles, num_periods)
        self.human_expert = HumanExpert()
        
        # Partnership metrics
        self.trust_metrics = TrustMetrics(
            overall_trust=0.5,
            prediction_accuracy=0.5,
            reliability_score=0.5,
            agreement_rate=0.5,
            learning_progress=0.0,
            relationship_strength=0.3
        )
        
        # Decision history
        self.partnership_history = []
        self.performance_timeline = []
        
        print(f"Initialized Human-AI Partnership for {len(customers)} customers, {len(vehicles)} vehicles")
    
    def collaborative_decision_making(self, period: int) -> Dict[str, Any]:
        """Execute collaborative decision-making process"""
        print(f"\n=== COLLABORATIVE DECISION-MAKING: Period {period + 1} ===")
        
        period_decisions = []
        period_performance = {
            'ai_recommendations': 0,
            'human_modifications': 0,
            'agreements': 0,
            'disagreements': 0,
            'avg_confidence': 0,
            'decision_quality': 0
        }
        
        # Phase 1: AI generates recommendations
        print("\nPhase 1: AI Generating Recommendations...")
        recommendations = self._generate_ai_recommendations(period)
        period_performance['ai_recommendations'] = len(recommendations)
        
        # Phase 2: Human expert reviews and provides feedback
        print("\nPhase 2: Human Expert Review...")
        feedback_list = []
        total_confidence = 0
        
        for i, recommendation in enumerate(recommendations):
            print(f"\n  Recommendation {i+1}: {recommendation.recommendation_type.upper()}")
            print(f"    AI Action: {recommendation.action}")
            print(f"    AI Confidence: {recommendation.confidence:.2f}")
            print(f"    AI Reasoning: {recommendation.reasoning[0]}")
            
            # Human expert review
            feedback = self.human_expert.review_recommendation(recommendation)
            feedback_list.append(feedback)
            total_confidence += feedback.confidence_in_decision
            
            print(f"    Human Decision: {'ACCEPT' if feedback.accepted else 'MODIFY'}")
            print(f"    Human Confidence: {feedback.confidence_in_decision:.2f}")
            print(f"    Human Reasoning: {feedback.reasoning}")
            
            # Update partnership metrics
            if feedback.accepted:
                period_performance['agreements'] += 1
            else:
                period_performance['disagreements'] += 1
                period_performance['human_modifications'] += 1
            
            # AI learns from feedback
            self.ai_system.learn_from_feedback(feedback)
            
            period_decisions.append({
                'recommendation': recommendation,
                'feedback': feedback,
                'final_action': self._implement_decision(recommendation, feedback)
            })
        
        # Calculate period metrics
        if feedback_list:
            period_performance['avg_confidence'] = total_confidence / len(feedback_list)
            period_performance['decision_quality'] = self._calculate_decision_quality(feedback_list)
        
        # Phase 3: Update trust and relationship metrics
        self._update_trust_metrics(period_performance)
        
        # Store period results
        self.partnership_history.append({
            'period': period,
            'decisions': period_decisions,
            'performance': period_performance
        })
        
        self.performance_timeline.append(period_performance)
        
        # Period summary
        print(f"\nPeriod {period + 1} Summary:")
        print(f"  AI Recommendations: {period_performance['ai_recommendations']}")
        print(f"  Human Modifications: {period_performance['human_modifications']}")
        print(f"  Agreement Rate: {period_performance['agreements']/(period_performance['agreements']+period_performance['disagreements'])*100:.1f}%")
        print(f"  Average Confidence: {period_performance['avg_confidence']:.2f}")
        print(f"  Decision Quality: {period_performance['decision_quality']:.2f}")
        print(f"  Overall Trust: {self.trust_metrics.overall_trust:.2f}")
        
        return period_performance
    
    def _generate_ai_recommendations(self, period: int) -> List[AIRecommendation]:
        """Generate AI recommendations for the period"""
        recommendations = []
        
        # Generate route recommendations for each vehicle
        for vehicle in self.vehicles[:2]:  # Limit to 2 vehicles for demo
            route_rec = self.ai_system.generate_route_recommendation(period, vehicle.id)
            recommendations.append(route_rec)
        
        # Generate inventory recommendations for urgent customers
        urgent_customers = self.ai_system._identify_urgent_customers()
        for customer_id in urgent_customers[:3]:  # Limit to 3 customers for demo
            inventory_rec = self.ai_system.generate_inventory_recommendation(period, customer_id)
            recommendations.append(inventory_rec)
        
        return recommendations
    
    def _implement_decision(self, recommendation: AIRecommendation, feedback: HumanFeedback) -> str:
        """Implement the final decision based on recommendation and feedback"""
        if feedback.accepted:
            return f"Implemented: {recommendation.action}"
        else:
            modifications_str = ", ".join([f"{k}: {v}" for k, v in feedback.modifications.items()])
            return f"Modified: {recommendation.action} | {modifications_str}"
    
    def _calculate_decision_quality(self, feedback_list: List[HumanFeedback]) -> float:
        """Calculate overall decision quality for the period"""
        if not feedback_list:
            return 0.5
        
        # Quality based on confidence and agreement
        avg_confidence = np.mean([f.confidence_in_decision for f in feedback_list])
        agreement_rate = sum(1 for f in feedback_list if f.accepted) / len(feedback_list)
        
        # Weight confidence more heavily
        quality = avg_confidence * 0.7 + agreement_rate * 0.3
        return quality
    
    def _update_trust_metrics(self, period_performance: Dict):
        """Update trust and relationship metrics"""
        # Calculate agreement rate
        total_decisions = period_performance['agreements'] + period_performance['disagreements']
        if total_decisions > 0:
            current_agreement_rate = period_performance['agreements'] / total_decisions
        else:
            current_agreement_rate = 0.5
        
        # Update agreement rate (exponential moving average)
        self.trust_metrics.agreement_rate = (
            0.7 * self.trust_metrics.agreement_rate + 0.3 * current_agreement_rate
        )
        
        # Update overall trust based on agreement and confidence
        trust_factor = (current_agreement_rate + period_performance['avg_confidence']) / 2
        self.trust_metrics.overall_trust = (
            0.8 * self.trust_metrics.overall_trust + 0.2 * trust_factor
        )
        
        # Update reliability score
        reliability = period_performance['decision_quality']
        self.trust_metrics.reliability_score = (
            0.8 * self.trust_metrics.reliability_score + 0.2 * reliability
        )
        
        # Update learning progress
        if len(self.performance_timeline) > 1:
            improvement = period_performance['avg_confidence'] - self.performance_timeline[-2]['avg_confidence']
            self.trust_metrics.learning_progress = max(0, self.trust_metrics.learning_progress + improvement * 0.1)
        
        # Update relationship strength
        self.trust_metrics.relationship_strength = (
            self.trust_metrics.overall_trust * 0.4 +
            self.trust_metrics.agreement_rate * 0.3 +
            self.trust_metrics.learning_progress * 0.3
        )
    
    def run_partnership_simulation(self) -> Dict[str, Any]:
        """Run the complete Human-AI Partnership simulation"""
        print(f"\n=== HUMAN-AI PARTNERSHIP SIMULATION ===")
        print(f"Simulating {self.num_periods} periods of collaborative decision-making...")
        
        for period in range(self.num_periods):
            self.collaborative_decision_making(period)
            
            # Simulate time progression
            time.sleep(0.1)
        
        # Calculate final results
        final_results = self._calculate_final_results()
        
        print(f"\n=== PARTNERSHIP SIMULATION COMPLETED ===")
        print(f"Final Trust Metrics:")
        print(f"  Overall Trust: {self.trust_metrics.overall_trust:.2f}")
        print(f"  Agreement Rate: {self.trust_metrics.agreement_rate:.2f}")
        print(f"  Reliability Score: {self.trust_metrics.reliability_score:.2f}")
        print(f"  Learning Progress: {self.trust_metrics.learning_progress:.2f}")
        print(f"  Relationship Strength: {self.trust_metrics.relationship_strength:.2f}")
        
        return final_results
    
    def _calculate_final_results(self) -> Dict[str, Any]:
        """Calculate final partnership results"""
        if not self.performance_timeline:
            return {}
        
        # Aggregate metrics
        total_recommendations = sum(p['ai_recommendations'] for p in self.performance_timeline)
        total_modifications = sum(p['human_modifications'] for p in self.performance_timeline)
        total_agreements = sum(p['agreements'] for p in self.performance_timeline)
        total_disagreements = sum(p['disagreements'] for p in self.performance_timeline)
        
        avg_confidence = np.mean([p['avg_confidence'] for p in self.performance_timeline])
        avg_quality = np.mean([p['decision_quality'] for p in self.performance_timeline])
        
        overall_agreement_rate = total_agreements / (total_agreements + total_disagreements) if (total_agreements + total_disagreements) > 0 else 0
        
        return {
            'total_recommendations': total_recommendations,
            'total_modifications': total_modifications,
            'total_agreements': total_agreements,
            'total_disagreements': total_disagreements,
            'overall_agreement_rate': overall_agreement_rate,
            'avg_confidence': avg_confidence,
            'avg_decision_quality': avg_quality,
            'final_trust_metrics': self.trust_metrics,
            'learning_trajectory': [p['avg_confidence'] for p in self.performance_timeline]
        }
```

```python
# Analyze Human-AI Partnership performance
def analyze_partnership_performance(partnership, results):
    """Analyze Human-AI Partnership performance and relationship development"""
    print("\n=== HUMAN-AI PARTNERSHIP ANALYSIS ===")
    
    # Trust development analysis
    print(f"\nTrust Development Analysis:")
    trust_metrics = results['final_trust_metrics']
    
    print(f"  Overall Trust: {trust_metrics.overall_trust:.2f} (Initial: 0.50)")
    print(f"  Agreement Rate: {trust_metrics.agreement_rate:.2f}")
    print(f"  Reliability Score: {trust_metrics.reliability_score:.2f}")
    print(f"  Learning Progress: {trust_metrics.learning_progress:.2f}")
    print(f"  Relationship Strength: {trust_metrics.relationship_strength:.2f}")
    
    # Decision quality analysis
    print(f"\nDecision Quality Analysis:")
    print(f"  Total Recommendations: {results['total_recommendations']}")
    print(f"  Human Modifications: {results['total_modifications']}")
    print(f"  Agreements: {results['total_agreements']}")
    print(f"  Disagreements: {results['total_disagreements']}")
    print(f"  Overall Agreement Rate: {results['overall_agreement_rate']:.1%}")
    print(f"  Average Confidence: {results['avg_confidence']:.2f}")
    print(f"  Average Decision Quality: {results['avg_decision_quality']:.2f}")
    
    # Learning trajectory analysis
    learning_trajectory = results['learning_trajectory']
    if len(learning_trajectory) > 1:
        initial_confidence = learning_trajectory[0]
        final_confidence = learning_trajectory[-1]
        improvement = final_confidence - initial_confidence
        
        print(f"\nLearning Trajectory:")
        print(f"  Initial Confidence: {initial_confidence:.2f}")
        print(f"  Final Confidence: {final_confidence:.2f}")
        print(f"  Improvement: {improvement:+.2f} ({improvement/initial_confidence*100:+.1f}%)")
        print(f"  Learning Rate: {improvement/len(learning_trajectory):.3f} per period")
    
    # Expert knowledge capture
    print(f"\nExpert Knowledge Capture:")
    expert_patterns = partnership.ai_system.expert_patterns
    print(f"  Captured Expert Patterns: {len(expert_patterns)}")
    
    if expert_patterns:
        accepted_patterns = [p for p in expert_patterns if p['context'] == 'accepted']
        print(f"  Accepted Patterns: {len(accepted_patterns)}")
        print(f"  Pattern Types: {list(set(p['context'] for p in expert_patterns))}")
    
    # AI system learning
    print(f"\nAI System Learning:")
    decision_history = partnership.ai_system.decision_history
    performance_history = partnership.ai_system.performance_history
    
    print(f"  Total Decisions Processed: {len(decision_history)}")
    print(f"  Performance History Points: {len(performance_history)}")
    
    if performance_history:
        recent_performance = performance_history[-5:]  # Last 5 periods
        avg_recent_performance = np.mean(recent_performance)
        print(f"  Recent Average Performance: {avg_recent_performance:.2f}")
    
    # Relationship effectiveness
    print(f"\nRelationship Effectiveness:")
    
    # Calculate collaboration efficiency
    collaboration_efficiency = results['overall_agreement_rate'] * results['avg_confidence']
    print(f"  Collaboration Efficiency: {collaboration_efficiency:.2f}")
    
    # Calculate knowledge transfer effectiveness
    knowledge_transfer = trust_metrics.learning_progress * trust_metrics.relationship_strength
    print(f"  Knowledge Transfer Effectiveness: {knowledge_transfer:.2f}")
    
    # Calculate decision improvement
    decision_improvement = trust_metrics.learning_progress * 100
    print(f"  Decision Improvement: {decision_improvement:.1f}%")
    
    return {
        'trust_development': trust_metrics.overall_trust - 0.5,  # Improvement from initial
        'agreement_rate': results['overall_agreement_rate'],
        'learning_improvement': trust_metrics.learning_progress,
        'collaboration_efficiency': collaboration_efficiency,
        'knowledge_transfer': knowledge_transfer,
        'decision_improvement': decision_improvement
    }

# Analyze performance
performance_analysis = analyze_partnership_performance(partnership, results)
```
