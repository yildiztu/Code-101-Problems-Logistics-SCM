# 94. The Multi-Echelon Inventory Optimization Problem

## Tier 8: The Value-Aligned & Ethical Framework (Equitable Service Levels)

### Key assumptions
- Pure cost optimization can have unintended negative social consequences
- Ethical constraints must be embedded directly into optimization logic
- Service level equity across different demographic groups is essential
- Multi-stakeholder utility optimization balances competing interests
- Constitutional AI principles guide decision-making processes

### Approach (step-by-step)
1. **Stakeholder Identification**: Map all affected groups and their priorities
2. **Ethical Principle Definition**: Establish constitutional AI principles for fairness
3. **Equity Constraints**: Define minimum service levels for essential goods
4. **Multi-Objective Optimization**: Balance cost, service, and equity objectives
5. **Value-Aligned Scoring**: Evaluate solutions against ethical criteria
6. **Transparency & Accountability**: Ensure explainable and auditable decisions
7. **Continuous Monitoring**: Track equity metrics and social impact

### What to look for in the results
- Service level equity across different regions and demographics
- Trade-offs between cost optimization and ethical constraints
- Impact on vulnerable populations and underserved areas
- Transparency of decision-making processes
- Long-term sustainability and social responsibility metrics

### Concrete example (from the source)
Ethical Service Level Constraints:
- **Essential medicines**: Minimum 99% service level for all regions
- **Baby formula**: Minimum 99% service level regardless of location profitability
- **Basic food staples**: Minimum 95% service level in rural and low-income areas
- **Cost Impact**: 15% increase in total network holding cost required
- **CEO Justification**: Ethical governance layer prevents service discrimination
- **Implementation**: Penalty functions enforce equity in optimization objective

### Visualization(s)
- Ethical governance architecture diagram
- Service level equity analysis across demographics
- Cost vs equity trade-off curves
- Stakeholder utility optimization results
- Social impact assessment dashboard

### Why this Tier exists vs earlier Tiers
Tier 8 addresses the critical limitation of pure cost optimization by embedding ethical principles directly into the optimization logic, ensuring that inventory decisions consider social equity and stakeholder welfare rather than just financial metrics.

### Pros / Cons vs earlier Tiers
**Pros vs Tiers 1-7:**
- Ensures equitable service levels across all demographics
- Incorporates social responsibility and ethical considerations
- Provides transparent and accountable decision-making
- Balances multiple stakeholder interests
- Prevents discriminatory optimization outcomes

**Pros:**

**Cons:**
- Higher operational costs due to equity constraints

### When to use this Tier

- Organizations operating in diverse demographic regions

```python
from typing import Tuple, List, Dict, Set

# Import required libraries for Ethical Framework implementation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass, field
from typing import List, Dict, Tuple, Optional, Any
import random
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Set plotting style for professional visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
@dataclass
class StakeholderGroup:
    """Represents a stakeholder group in the supply chain"""
    group_id: str
    name: str
    group_type: str  # 'demographic', 'geographic', 'economic', 'vulnerable'
    priority: float  # 0.0 to 1.0, higher = more priority
    locations: List[str]
    service_requirements: Dict[str, float]  # sku -> minimum service level
    population_served: int
    vulnerability_score: float  # 0.0 to 1.0, higher = more vulnerable

@dataclass
class EthicalPrinciple:
    """Represents an ethical principle for optimization"""
    principle_id: str
    name: str
    description: str
    weight: float  # Importance weight in optimization
    constraint_type: str  # 'hard', 'soft', 'penalty'
    implementation: str  # How the principle is enforced

@dataclass
class EquityConstraint:
    """Represents an equity constraint in optimization"""
    constraint_id: str
    name: str
    stakeholder_group: str
    sku_category: str  # 'essential', 'important', 'standard'
    minimum_service_level: float
    penalty_weight: float
    enforcement_type: str  # 'hard_constraint', 'penalty_function'

@dataclass
class EthicalSolution:
    """Represents an ethically evaluated inventory solution"""
    solution_id: str
    inventory_levels: Dict[str, float]
    total_cost: float
    service_levels: Dict[str, float]
    equity_score: float
    ethical_compliance: Dict[str, bool]
    stakeholder_utility: Dict[str, float]
    transparency_report: Dict[str, Any]
```

```python
try:
    class EthicalOptimizationEngine:
        """Ethical optimization engine for multi-echelon inventory"""
        
        def __init__(self, network_config: Dict):
            self.network_config = network_config
            self.stakeholder_groups = {}
            self.ethical_principles = {}
            self.equity_constraints = {}
            self.optimization_history = []
            
            self._initialize_ethical_framework()
        
        def _initialize_ethical_framework(self):
            """Initialize ethical principles and stakeholder groups"""
            
            # Define stakeholder groups
            self.stakeholder_groups = {
                'urban_wealthy': StakeholderGroup(
                    group_id='urban_wealthy',
                    name='Urban High-Income Areas',
                    group_type='economic',
                    priority=0.6,
                    locations=['DC_Urban_North', 'DC_Urban_South'],
                    service_requirements={'essential': 0.95, 'standard': 0.90},
                    population_served=500000,
                    vulnerability_score=0.2
                ),
                'urban_middle': StakeholderGroup(
                    group_id='urban_middle',
                    name='Urban Middle-Income Areas',
                    group_type='economic',
                    priority=0.8,
                    locations=['DC_Urban_Central', 'DC_Urban_East'],
                    service_requirements={'essential': 0.97, 'standard': 0.92},
                    population_served=800000,
                    vulnerability_score=0.4
                ),
                'rural_general': StakeholderGroup(
                    group_id='rural_general',
                    name='Rural Communities',
                    group_type='geographic',
                    priority=0.9,
                    locations=['DC_Rural_West', 'DC_Rural_East'],
                    service_requirements={'essential': 0.98, 'standard': 0.93},
                    population_served=200000,
                    vulnerability_score=0.7
                ),
                'vulnerable_elderly': StakeholderGroup(
                    group_id='vulnerable_elderly',
                    name='Elderly Care Facilities',
                    group_type='vulnerable',
                    priority=1.0,
                    locations=['DC_Healthcare_North', 'DC_Healthcare_South'],
                    service_requirements={'essential': 0.99, 'standard': 0.95},
                    population_served=50000,
                    vulnerability_score=0.9
                ),
                'vulnerable_children': StakeholderGroup(
                    group_id='vulnerable_children',
                    name='Schools & Pediatric Care',
                    group_type='vulnerable',
                    priority=1.0,
                    locations=['DC_Education_Central', 'DC_Education_West'],
                    service_requirements={'essential': 0.99, 'standard': 0.95},
                    population_served=100000,
                    vulnerability_score=0.95
                )
            }
            
            # Define ethical principles
            self.ethical_principles = {
                'equity': EthicalPrinciple(
                    principle_id='equity',
                    name='Service Level Equity',
                    description='Ensure equitable service levels across all demographic groups',
                    weight=0.3,
                    constraint_type='hard',
                    implementation='minimum_service_constraints'
                ),
                'priority_vulnerable': EthicalPrinciple(
                    principle_id='priority_vulnerable',
                    name='Priority for Vulnerable Populations',
                    description='Give priority service to vulnerable and high-need groups',
                    weight=0.25,
                    constraint_type='hard',
                    implementation='enhanced_service_requirements'
                ),
                'transparency': EthicalPrinciple(
                    principle_id='transparency',
                    name='Decision Transparency',
                    description='Ensure all decisions are explainable and auditable',
                    weight=0.2,
                    constraint_type='soft',
                    implementation='explainable_optimization'
                ),
                'sustainability': EthicalPrinciple(
                    principle_id='sustainability',
                    name='Long-term Sustainability',
                    description='Balance short-term costs with long-term social sustainability',
                    weight=0.15,
                    constraint_type='soft',
                    implementation='multi_objective_optimization'
                ),
                'accountability': EthicalPrinciple(
                    principle_id='accountability',
                    name='Stakeholder Accountability',
                    description='Ensure accountability to all stakeholders for service delivery',
                    weight=0.1,
                    constraint_type='soft',
                    implementation='performance_tracking'
                )
            }
            
            # Define equity constraints
            self.equity_constraints = {
                'essential_medicine': EquityConstraint(
                    constraint_id='essential_medicine',
                    name='Essential Medicine Service Level',
                    stakeholder_group='all',
                    sku_category='essential',
                    minimum_service_level=0.99,
                    penalty_weight=1000.0,
                    enforcement_type='hard_constraint'
                ),
                'baby_formula': EquityConstraint(
                    constraint_id='baby_formula',
                    name='Baby Formula Availability',
                    stakeholder_group='all',
                    sku_category='essential',
                    minimum_service_level=0.99,
                    penalty_weight=1000.0,
                    enforcement_type='hard_constraint'
                ),
                'basic_food': EquityConstraint(
                    constraint_id='basic_food',
                    name='Basic Food Staples',
                    stakeholder_group='rural_general',
                    sku_category='essential',
                    minimum_service_level=0.95,
                    penalty_weight=500.0,
                    enforcement_type='penalty_function'
                ),
                'vulnerable_priority': EquityConstraint(
                    constraint_id='vulnerable_priority',
                    name='Vulnerable Group Priority',
                    stakeholder_group='vulnerable',
                    sku_category='essential',
                    minimum_service_level=0.99,
                    penalty_weight=750.0,
                    enforcement_type='hard_constraint'
                )
            }
        
        def calculate_equity_penalty(self, service_levels: Dict[str, float], 
                                    inventory_levels: Dict[str, float]) -> float:
            """Calculate penalty for equity violations"""
            
            total_penalty = 0.0
            
            for constraint_id, constraint in self.equity_constraints.items():
                
                if constraint.stakeholder_group == 'all':
                    # Apply to all stakeholder groups
                    for group_id, group in self.stakeholder_groups.items():
                        for location in group.locations:
                            if location in service_levels:
                                current_service = service_levels[location]
                                
                                if current_service < constraint.minimum_service_level:
                                    # Calculate penalty based on violation severity
                                    violation = constraint.minimum_service_level - current_service
                                    penalty = violation * constraint.penalty_weight
                                    total_penalty += penalty
                else:
                    # Apply to specific stakeholder group
                    if constraint.stakeholder_group in self.stakeholder_groups:
                        group = self.stakeholder_groups[constraint.stakeholder_group]
                        for location in group.locations:
                            if location in service_levels:
                                current_service = service_levels[location]
                                
                                if current_service < constraint.minimum_service_level:
                                    violation = constraint.minimum_service_level - current_service
                                    penalty = violation * constraint.penalty_weight
                                    total_penalty += penalty
            
            return total_penalty
        
        def calculate_stakeholder_utility(self, service_levels: Dict[str, float], 
                                         inventory_levels: Dict[str, float]) -> Dict[str, float]:
            """Calculate utility for each stakeholder group"""
            
            utilities = {}
            
            for group_id, group in self.stakeholder_groups.items():
                
                # Calculate average service level for this group
                group_services = []
                for location in group.locations:
                    if location in service_levels:
                        group_services.append(service_levels[location])
                
                if group_services:
                    avg_service = np.mean(group_services)
                    
                    # Calculate utility based on service level and priority
                    # Higher priority groups get more utility from good service
                    base_utility = avg_service
                    priority_multiplier = 1 + (group.priority - 0.5)  # 0.5 to 1.5 multiplier
                    vulnerability_bonus = group.vulnerability_score * 0.2  # Bonus for serving vulnerable
                    
                    utility = base_utility * priority_multiplier + vulnerability_bonus
                    utilities[group_id] = min(1.0, utility)  # Cap at 1.0
                else:
                    utilities[group_id] = 0.0
            
            return utilities
        
        def optimize_ethical_inventory(self, base_demand: Dict[str, float], 
                                         base_costs: Dict[str, float]) -> EthicalSolution:
            """Optimize inventory levels with ethical constraints"""
            
            # Simplified ethical optimization using weighted approach
            locations = list(base_demand.keys())
            
            # Start with cost-optimal solution (simplified)
            optimal_inventory = {}
            service_levels = {}
            
            for location in locations:
                demand = base_demand[location]
                cost = base_costs[location]
                
                # Find which stakeholder group this location belongs to
                group_id = self._get_location_stakeholder(location)
                group = self.stakeholder_groups.get(group_id)
                
                if group:
                    # Apply ethical adjustments
                    base_stock = demand * 1.5  # Base safety stock
                    
                    # Adjust for vulnerability and priority
                    if group.vulnerability_score > 0.7:
                        base_stock *= 1.3  # Higher stock for vulnerable groups
                    
                    if group.priority > 0.8:
                        base_stock *= 1.2  # Higher stock for priority groups
                    
                    optimal_inventory[location] = base_stock
                    
                    # Calculate service level (simplified)
                    service_levels[location] = min(0.99, base_stock / (demand * 2))
                else:
                    optimal_inventory[location] = demand * 1.5
                    service_levels[location] = 0.90
            
            # Calculate total cost
            total_cost = sum(optimal_inventory[loc] * base_costs[loc] for loc in locations)
            
            # Calculate equity penalty
            equity_penalty = self.calculate_equity_penalty(service_levels, optimal_inventory)
            total_cost += equity_penalty
            
            # Calculate stakeholder utility
            stakeholder_utility = self.calculate_stakeholder_utility(service_levels, optimal_inventory)
            
            # Calculate equity score
            equity_score = self._calculate_equity_score(service_levels)
            
            # Check ethical compliance
            ethical_compliance = self._check_ethical_compliance(service_levels)
            
            # Generate transparency report
            transparency_report = self._generate_transparency_report(
                optimal_inventory, service_levels, total_cost, equity_penalty
            )
            
            return EthicalSolution(
                solution_id=f"ETH_{datetime.now().strftime('%Y%m%d_%H%M%S')}",
                inventory_levels=optimal_inventory,
                total_cost=total_cost,
                service_levels=service_levels,
                equity_score=equity_score,
                ethical_compliance=ethical_compliance,
                stakeholder_utility=stakeholder_utility,
                transparency_report=transparency_report
            )
        
        def _get_location_stakeholder(self, location: str) -> str:
            """Get stakeholder group for a location"""
            for group_id, group in self.stakeholder_groups.items():
                if location in group.locations:
                    return group_id
            return 'urban_middle'  # Default
        
        def _calculate_equity_score(self, service_levels: Dict[str, float]) -> float:
            """Calculate overall equity score"""
            if not service_levels:
                return 0.0
            
            # Calculate service level variance across groups
            group_services = {}
            
            for group_id, group in self.stakeholder_groups.items():
                group_levels = []
                for location in group.locations:
                    if location in service_levels:
                        group_levels.append(service_levels[location])
                
                if group_levels:
                    group_services[group_id] = np.mean(group_levels)
            
            if group_services:
                services = list(group_services.values())
                # Lower variance = higher equity
                variance = np.var(services)
                equity_score = max(0.0, 1.0 - variance * 10)  # Scale variance penalty
                return equity_score
            
            return 0.0
        
        def _check_ethical_compliance(self, service_levels: Dict[str, float]) -> Dict[str, bool]:
            """Check compliance with ethical constraints"""
            compliance = {}
            
            for constraint_id, constraint in self.equity_constraints.items():
                compliant = True
                
                if constraint.stakeholder_group == 'all':
                    # Check all locations
                    for location in service_levels:
                        if service_levels[location] < constraint.minimum_service_level:
                            compliant = False
                            break
                else:
                    # Check specific stakeholder group
                    if constraint.stakeholder_group in self.stakeholder_groups:
                        group = self.stakeholder_groups[constraint.stakeholder_group]
                        for location in group.locations:
                            if location in service_levels:
                                if service_levels[location] < constraint.minimum_service_level:
                                    compliant = False
                                    break
                
                compliance[constraint_id] = compliant
            
            return compliance
        
        def _generate_transparency_report(self, inventory: Dict[str, float], 
                                         service: Dict[str, float], 
                                         cost: float, penalty: float) -> Dict[str, Any]:
            """Generate transparency report for the solution"""
            
            return {
                'optimization_method': 'ethical_multi_objective',
                'total_cost_breakdown': {
                    'holding_cost': cost - penalty,
                    'equity_penalty': penalty,
                    'total_cost': cost
                },
                'service_level_summary': {
                    'average_service': np.mean(list(service.values())),
                    'min_service': np.min(list(service.values())),
                    'max_service': np.max(list(service.values())),
                    'service_variance': np.var(list(service.values()))
                },
                'ethical_principles_applied': list(self.ethical_principles.keys()),
                'stakeholder_impact': self._calculate_stakeholder_utility(service, inventory),
                'decision_rationale': self._generate_decision_rationale()
                'audit_trail': f"Optimization performed at {datetime.now().isoformat()}"
            }
        
        def _generate_decision_rationale(self) -> List[str]:
            """Generate rationale for ethical decisions"""
            rationale = [
                "Optimization prioritized equitable service levels across all demographic groups",",
                "Higher inventory levels allocated to vulnerable populations per ethical principles",",
                "Cost-benefit analysis balanced with social welfare considerations",",
                "Minimum service level constraints enforced for essential goods",",
                "Stakeholder utility maximization weighted by vulnerability and priority"
            ]
            return rationale
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    # Compare with traditional cost-only optimization
    def traditional_cost_optimization(demand: Dict[str, float], costs: Dict[str, float]) -> Dict[str, float]:
        """Traditional cost-only optimization for comparison"""
        optimal_inventory = {}
        service_levels = {}
        
        for location in demand:
            # Simple cost-optimal: minimize inventory while meeting basic service
            base_stock = demand[location] * 1.2  # Minimal safety stock
            optimal_inventory[location] = base_stock
            service_levels[location] = min(0.90, base_stock / (demand[location] * 1.5))
        
        total_cost = sum(optimal_inventory[loc] * costs[loc] for loc in demand)
        
        return {
            'inventory': optimal_inventory,
            'service_levels': service_levels,
            'total_cost': total_cost
        }
    
    # Run traditional optimization
    traditional_result = traditional_cost_optimization(base_demand, base_costs)
    
    # Calculate traditional equity metrics
    traditional_equity_penalty = ethical_engine.calculate_equity_penalty(
        traditional_result['service_levels'], traditional_result['inventory']
    )
    traditional_equity_score = ethical_engine._calculate_equity_score(traditional_result['service_levels'])
    traditional_utility = ethical_engine.calculate_stakeholder_utility(
        traditional_result['service_levels'], traditional_result['inventory']
    )
    
    print("\n" + "=" * 60)
    print("COMPARISON: ETHICAL VS TRADITIONAL OPTIMIZATION")
    print("=" * 60)
    print()
    
    print("Cost Comparison:")
    print("-" * 20)
    print(f"Traditional Total Cost: ${traditional_result['total_cost']:.2f}")
    print(f"Ethical Total Cost:    ${ethical_solution.total_cost:.2f}")
    cost_increase = (ethical_solution.total_cost - traditional_result['total_cost']) / traditional_result['total_cost'] * 100
    print(f"Cost Increase:        {cost_increase:.1f}%")
    print()
    
    print("Equity Comparison:")
    print("-" * 20)
    print(f"Traditional Equity Score: {traditional_equity_score:.3f}")
    print(f"Ethical Equity Score:    {ethical_solution.equity_score:.3f}")
    equity_improvement = (ethical_solution.equity_score - traditional_equity_score) / traditional_equity_score * 100 if traditional_equity_score > 0 else 0
    print(f"Equity Improvement:     {equity_improvement:.1f}%")
    print()
    
    print("Service Level Comparison:")
    print("-" * 30)
    traditional_avg_service = np.mean(list(traditional_result['service_levels'].values()))
    ethical_avg_service = np.mean(list(ethical_solution.service_levels.values()))
    print(f"Traditional Avg Service: {traditional_avg_service:.1%}")
    print(f"Ethical Avg Service:    {ethical_avg_service:.1%}")
    service_improvement = (ethical_avg_service - traditional_avg_service) / traditional_avg_service * 100
    print(f"Service Improvement:    {service_improvement:.1f}%")
    print()
    
    print("Stakeholder Utility Comparison:")
    print("-" * 35)
    print(f"{'Stakeholder Group':<20} {'Traditional':<12} {'Ethical':<10} {'Improvement':<12}")
    print("-" * 55)
    for group_id in ethical_engine.stakeholder_groups:
        group_name = ethical_engine.stakeholder_groups[group_id].name
        trad_utility = traditional_utility.get(group_id, 0.0)
        eth_utility = ethical_solution.stakeholder_utility.get(group_id, 0.0)
        improvement = ((eth_utility - trad_utility) / trad_utility * 100) if trad_utility > 0 else 0
        print(f"{group_name[:19]:<20} {trad_utility:<12.3f} {eth_utility:<10.3f} {improvement:<12.1f}%")
except Exception as e:
    print(f'Error in cell: {e}')
```
