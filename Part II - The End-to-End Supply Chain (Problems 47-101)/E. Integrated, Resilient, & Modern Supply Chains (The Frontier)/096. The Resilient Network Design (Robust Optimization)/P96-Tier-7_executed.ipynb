# 96. The Resilient Network Design (Robust Optimization)
## Tier 7: The Human-AI Symbiotic Partnership (The Centaur Model)

### Key assumptions
- Human expertise and AI capabilities can create synergistic value beyond either alone
- Explainable AI provides transparency for human decision-makers
- Trust-based relationships develop through collaborative problem-solving
- Human contextual understanding complements AI computational power
- Iterative refinement improves both human and AI performance over time

### Approach (step-by-step)
1. **Define human-AI collaboration framework** with clear roles and responsibilities
2. **Implement explainable AI system** with transparent reasoning processes
3. **Create trust development mechanism** through successful collaboration
4. **Establish human-in-the-loop decision process** for critical choices
5. **Design learning system** where both human and AI improve from experience
6. **Measure partnership effectiveness** through objective and subjective metrics

### What to look for in the results
- Trust development and agreement rates between human and AI
- Decision quality improvement through collaboration
- Explainable AI reasoning and transparency metrics
- Performance comparison with pure AI and pure human approaches

### Concrete example (from the source)
Human-AI partnership for southeastern US network redesign:
- 5 human experts + 3 AI agents collaborative decision-making
- 94% agreement rate achieved through iterative refinement
- 15% improvement over pure AI optimization
- 8% higher resilience than pure AI solution
- 23% lower costs than original human-only design

```python
try:
    # Define data structures for Human-AI Partnership
    @dataclass
    class Facility:
        """Represents a facility in the network"""
        id: int
        x: float
        y: float
        capacity: int
        fixed_cost: float
        operational_status: bool = True
        strategic_importance: str = "medium"  # low, medium, high, critical
        local_constraints: List[str] = None  # Regulatory, environmental, etc.
    
    @dataclass
    class DemandZone:
        """Represents a demand zone with contextual factors"""
        id: int
        x: float
        y: float
        base_demand: float
        population_served: int = 0
        vulnerability_level: str = "medium"  # low, medium, high, critical
        access_constraints: List[str] = None  # Transportation, geographic, etc.
    
    @dataclass
    class HumanExpert:
        """Represents a human expert in the partnership"""
        id: int
        name: str
        expertise_area: str  # logistics, regulations, local_knowledge, finance, operations
        experience_years: int
        trust_level: float = 0.5  # 0-1, develops over time
        decision_confidence: float = 0.7  # 0-1
        domain_knowledge: Dict[str, float] = None  # Area-specific knowledge scores
    
    @dataclass
    class AIAgent:
        """Represents an AI agent in the partnership"""
        id: int
        name: str
        capability: str  # optimization, forecasting, risk_analysis, pattern_recognition
        processing_power: float  # Relative computational capability
        explainability_score: float = 0.8  # 0-1, ability to explain reasoning
        accuracy_rate: float = 0.85  # 0-1, historical performance
        learning_rate: float = 0.1  # 0-1, improvement speed
    
    @dataclass
    class CollaborationSession:
        """Represents a human-AI collaboration session"""
        session_id: int
     timestamp: datetime
        participants: List[Union[HumanExpert, AIAgent]]
     problem_context: str
     ai_recommendation: Dict
    human_feedback: Dict
    final_decision: Dict
    agreement_level: float  # 0-1, consensus level
    trust_change: float  # Change in trust after this session
    
    @dataclass
    class PartnershipMetrics:
        """Metrics for evaluating human-AI partnership performance"""
        overall_agreement_rate: float
    decision_quality_score: float
    trust_development_score: float
    collaboration_efficiency: float
    learning_improvement_rate: float
    stakeholder_satisfaction: float
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    # Implement the Human-AI Symbiotic Partnership framework
    class HumanAISymbioticPartnership:
        """Human-AI symbiotic partnership for resilient network design"""
        
        def __init__(self, facilities: List[Facility], demand_zones: List[DemandZone],
                     human_experts: List[HumanExpert], ai_agents: List[AIAgent]):
            self.facilities = facilities
            self.demand_zones = demand_zones
            self.human_experts = human_experts
            self.ai_agents = ai_agents
            
            # Partnership state
            self.collaboration_history = []
            self.trust_matrix = np.zeros((len(human_experts), len(ai_agents)))
            self.partnership_metrics = None
            
            # Learning parameters
            self.trust_development_rate = 0.05
            self.learning_decay_rate = 0.95
            
            # Initialize trust matrix with moderate values
            for i in range(len(human_experts)):
                for j in range(len(ai_agents)):
                    self.trust_matrix[i, j] = 0.5  # Initial moderate trust
        
        def generate_ai_recommendation(self, problem_context: str) -> Dict:
            """Generate AI recommendation with explainable reasoning"""
            
            # Use optimization AI for network design
            opt_ai = next((ai for ai in self.ai_agents if ai.capability == "optimization"), None)
            
            # Generate facility selection using optimization
            selected_facilities = self._ai_optimize_network()
            
            # Generate explainable reasoning
            reasoning = self._generate_explainable_reasoning(selected_facilities, problem_context)
            
            # Calculate confidence and uncertainty
            confidence = self._calculate_ai_confidence(selected_facilities)
            uncertainty = self._calculate_uncertainty(selected_facilities)
            
            return {
                "selected_facilities": selected_facilities,
                "reasoning": reasoning,
                "confidence": confidence,
                "uncertainty": uncertainty,
                "expected_performance": self._evaluate_solution_performance(selected_facilities),
                "alternative_options": self._generate_alternatives(selected_facilities)
                "risk_factors": self._identify_risk_factors(selected_facilities)
            }
        
        def _ai_optimize_network(self) -> List[int]:
            """AI optimization for facility selection"""
            # Simplified optimization - select top facilities based on cost-benefit analysis
            facility_scores = []
            
            for facility in self.facilities:
                # Calculate cost-benefit score
                strategic_bonus = {"critical": 1.2, "high": 1.1, "medium": 1.0, "low": 0.9}
                
                # Calculate service coverage potential
                service_potential = 0
                for demand in self.demand_zones:
                    distance = np.sqrt((facility.x - demand.x)**2 + (facility.y - demand.y)**2)
                    service_potential += demand.base_demand * np.exp(-distance / 1000)
                
                score = (service_potential / facility.fixed_cost) * strategic_bonus.get(facility.strategic_importance, 1.0)
                facility_scores.append((score, facility.id))
            
            # Select top facilities (at least 3, at most 4)
            facility_scores.sort(reverse=True)
            selected = [fid for score, fid in facility_scores[:4]]
            
            return selected
        
        def _generate_explainable_reasoning(self, selected_facilities: List[int], context: str) -> List[str]:
            """Generate explainable reasoning for the AI recommendation"""
            
            reasoning = []
            
            # Primary reasoning - cost efficiency
            total_cost = sum(f.fixed_cost for f in self.facilities if f.id in selected_facilities)
            reasoning.append(f"Primary objective: Minimize total fixed cost while maintaining service coverage")
            reasoning.append(f"Selected facilities have total cost of ${total_cost:,}, representing optimal cost-benefit ratio")
            
            # Geographic distribution reasoning
            selected_fac_objs = [f for f in self.facilities if f.id in selected_facilities]
            center_x = np.mean([f.x for f in selected_fac_objs])
            center_y = np.mean([f.y for f in selected_fac_objs])
            
            max_distance = max([np.sqrt((f.x - center_x)**2 + (f.y - center_y)**2) for f in selected_fac_objs])
            reasoning.append(f"Geographic distribution: Facilities distributed around center point ({center_x:.1f}, {center_y:.1f})")
            reasoning.append(f"Maximum dispersion distance: {max_distance:.1f} units, ensuring geographical diversity")
            
            # Strategic importance reasoning
            critical_facilities = [f.id for f in selected_fac_objs if f.strategic_importance == "critical"]
            if critical_facilities:
                reasoning.append(f"Strategic coverage: Critical facilities {critical_facilities} included for network resilience")
            
            # Service coverage reasoning
            total_service_potential = 0
            for facility in selected_fac_objs:
                for demand in self.demand_zones:
                    distance = np.sqrt((facility.x - demand.x)**2 + (facility.y - demand.y)**2)
                    total_service_potential += demand.base_demand * np.exp(-distance / 1000)
            
            reasoning.append(f"Service coverage: Selected configuration can serve approximately {total_service_potential:.0f} demand units")
            
            # Constraint handling reasoning
            constraint_issues = []
            for facility in selected_fac_objs:
                if facility.local_constraints:
                    constraint_issues.extend([f"F{facility.id}: {', '.join(facility.local_constraints)}"])
            
            if constraint_issues:
                reasoning.append(f"Constraint considerations: {'; '.join(constraint_issues)}")
            else:
                reasoning.append("Constraint considerations: No major regulatory or operational constraints identified")
            
            return reasoning
        
        def _calculate_ai_confidence(self, selected_facilities: List[int]) -> float:
            """Calculate AI confidence in the recommendation"""
            
            # Base confidence from AI agent accuracy
            opt_ai = next((ai for ai in self.ai_agents if ai.capability == "optimization"), None)
            base_confidence = opt_ai.accuracy_rate if opt_ai else 0.8
            
            # Adjust based on solution quality
            performance = self._evaluate_solution_performance(selected_facilities)
            performance_factor = min(1.0, performance / 0.8)  # Normalize to 0.8 target
            
            # Adjust based on uncertainty
            uncertainty = self._calculate_uncertainty(selected_facilities)
            uncertainty_penalty = uncertainty * 0.2
            
            confidence = base_confidence * performance_factor - uncertainty_penalty
            return max(0.0, min(1.0, confidence))
        
        def _calculate_uncertainty(self, selected_facilities: List[int]) -> float:
            """Calculate uncertainty in the recommendation"""
            
            # Uncertainty from demand variability
            demand_variance = np.var([d.base_demand for d in self.demand_zones])
            demand_uncertainty = min(1.0, demand_variance / 10000)
            
            # Uncertainty from constraint complexity
            constraint_complexity = 0
            for facility in self.facilities:
                if facility.id in selected_facilities and facility.local_constraints:
                    constraint_complexity += len(facility.local_constraints)
            
            constraint_uncertainty = min(1.0, constraint_complexity / 10)
            
            # Uncertainty from geographic dispersion
            selected_fac_objs = [f for f in self.facilities if f.id in selected_facilities]
            if len(selected_fac_objs) > 1:
                distances = []
                for i in range(len(selected_fac_objs)):
                    for j in range(i + 1, len(selected_fac_objs)):
                        dist = np.sqrt((selected_fac_objs[i].x - selected_fac_objs[j].x)**2 + 
                                     (selected_fac_objs[i].y - selected_fac_objs[j].y)**2)
                        distances.append(dist)
                
                distance_variance = np.var(distances) if distances else 0
                geographic_uncertainty = min(1.0, distance_variance / 1000)
            else:
                geographic_uncertainty = 0.5
            
            # Combine uncertainties
            total_uncertainty = (demand_uncertainty + constraint_uncertainty + geographic_uncertainty) / 3
            
            return total_uncertainty
        
        def _evaluate_solution_performance(self, selected_facilities: List[int]) -> float:
            """Evaluate the performance of a facility selection"""
            
            selected_fac_objs = [f for f in self.facilities if f.id in selected_facilities]
            
            if len(selected_fac_objs) < 2:
                return 0.0
            
            # Calculate service coverage
            total_demand = sum(d.base_demand for d in self.demand_zones)
            served_demand = 0
            
            for demand in self.demand_zones:
                best_service = 0
                for facility in selected_fac_objs:
                    distance = np.sqrt((facility.x - demand.x)**2 + (facility.y - demand.y)**2)
                    service = min(facility.capacity, demand.base_demand) * np.exp(-distance / 1000)
                    best_service = max(best_service, service)
                served_demand += best_service
            
            service_coverage = served_demand / total_demand if total_demand > 0 else 0
            
            return service_coverage
        
        def _generate_alternatives(self, selected_facilities: List[int]) -> List[Dict]:
            """Generate alternative solutions for comparison"""
            
            alternatives = []
            
            # Alternative 1: Cost-focused (cheapest facilities)
            cost_sorted = sorted(self.facilities, key=lambda f: f.fixed_cost)
            cost_focused = [f.id for f in cost_sorted[:3]]
            
            alternatives.append({
                "name": "Cost-Focused",
                "facilities": cost_focused,
                "performance": self._evaluate_solution_performance(cost_focused),
                "total_cost": sum(f.fixed_cost for f in self.facilities if f.id in cost_focused)
            })
            
            # Alternative 2: Coverage-focused (best service coverage)
            best_coverage = []
            best_performance = 0
            
            for i in range(len(self.facilities)):
                for j in range(i + 1, len(self.facilities)):
                    for k in range(j + 1, len(self.facilities)):
                        test_selection = [self.facilities[i].id, self.facilities[j].id, self.facilities[k].id]
                        performance = self._evaluate_solution_performance(test_selection)
                        
                        if performance > best_performance:
                            best_performance = performance
                            best_coverage = test_selection
            
            alternatives.append({
                "name": "Coverage-Focused",
                "facilities": best_coverage,
                "performance": best_performance,
                "total_cost": sum(f.fixed_cost for f in self.facilities if f.id in best_coverage)
            })
            
            return alternatives
        
        def _identify_risk_factors(self, selected_facilities: List[int]) -> List[str]:
            """Identify potential risk factors in the solution"""
            
            risks = []
            selected_fac_objs = [f for f in self.facilities if f.id in selected_facilities]
            
            # Geographic concentration risk
            if len(selected_fac_objs) >= 2:
                center_x = np.mean([f.x for f in selected_fac_objs])
                center_y = np.mean([f.y for f in selected_fac_objs])
                
                max_distance = max([np.sqrt((f.x - center_x)**2 + (f.y - center_y)**2) for f in selected_fac_objs])
                if max_distance < 30:  # Too concentrated
                    risks.append("Geographic concentration: Facilities too close together, vulnerable to regional disruptions")
            
            # Constraint violation risk
            for facility in selected_fac_objs:
                if facility.local_constraints:
                    risks.append(f"Constraint risk for F{facility.id}: {', '.join(facility.local_constraints)}")
            
            # Single point of failure risk
            critical_count = sum(1 for f in selected_fac_objs if f.strategic_importance == "critical")
            if critical_count == 0:
                risks.append("No critical facilities selected: May impact strategic importance")
            elif critical_count == 1:
                risks.append("Single critical facility: Potential single point of failure")
            
            # Capacity risk
            total_capacity = sum(f.capacity for f in selected_fac_objs)
            total_demand = sum(d.base_demand for d in self.demand_zones)
            
            if total_capacity < total_demand * 0.8:
                risks.append("Capacity risk: Total capacity may be insufficient for peak demand")
            
            return risks
        
        def process_human_feedback(self, ai_recommendation: Dict, human_expert: HumanExpert, 
                                   problem_context: str) -> Dict:
            """Process human expert feedback on AI recommendation"""
            
            # Human evaluates AI recommendation based on expertise
            feedback = {}
            
            # Domain-specific evaluation
            if human_expert.expertise_area == "regulations":
                feedback = self._evaluate_regulatory_compliance(ai_recommendation, human_expert)
            elif human_expert.expertise_area == "local_knowledge":
                feedback = self._evaluate_local_factors(ai_recommendation, human_expert)
            elif human_expert.expertise_area == "finance":
                feedback = self._evaluate_financial_implications(ai_recommendation, human_expert)
            elif human_expert.expertise_area == "operations":
                feedback = self._evaluate_operational_feasibility(ai_recommendation, human_expert)
            else:  # logistics
                feedback = self._evaluate_logistics_optimization(ai_recommendation, human_expert)
            
            # Human confidence assessment
            feedback["confidence"] = self._assess_human_confidence(ai_recommendation, human_expert)
            
            # Human concerns and suggestions
            feedback["concerns"] = self._identify_human_concerns(ai_recommendation, human_expert)
            feedback["suggestions"] = self._generate_human_suggestions(ai_recommendation, human_expert)
            
            return feedback
        
        def _evaluate_regulatory_compliance(self, ai_recommendation: Dict, expert: HumanExpert) -> Dict:
            """Evaluate regulatory compliance aspects"""
            
            evaluation = {"compliance_score": 0.0, "violations": [], "recommendations": []}
            
            selected_facilities = [f for f in self.facilities if f.id in ai_recommendation["selected_facilities"]]
            
            for facility in selected_facilities:
                if facility.local_constraints:
                    for constraint in facility.local_constraints:
                        # Expert evaluates compliance based on knowledge
                        compliance_knowledge = expert.domain_knowledge.get(constraint.replace("_", ""), 0.8)
                        
                        if compliance_knowledge < 0.7:
                            evaluation["violations"].append(f"F{facility.id}: {constraint} compliance concern")
                            evaluation["compliance_score"] -= 0.2
                        else:
                            evaluation["compliance_score"] += 0.1
            
            # Generate compliance recommendations
            if evaluation["violations"]:
                evaluation["recommendations"].append("Conduct detailed regulatory review before implementation")
                evaluation["recommendations"].append("Consider alternative facilities with fewer constraints")
            else:
                evaluation["recommendations"].append("Regulatory compliance appears satisfactory")
            
            return evaluation
        
        def _evaluate_local_factors(self, ai_recommendation: Dict, expert: HumanExpert) -> Dict:
            """Evaluate local knowledge factors"""
            
            evaluation = {"local_score": 0.0, "issues": [], "insights": []}
            
            selected_facilities = [f for f in self.facilities if f.id in ai_recommendation["selected_facilities"]]
            
            # Evaluate community relations
            for facility in selected_facilities:
                local_knowledge = expert.domain_knowledge.get("community_relations", 0.8)
                
                # Check if facility is in sensitive area
                for demand in self.demand_zones:
                    distance = np.sqrt((facility.x - demand.x)**2 + (facility.y - demand.y)**2)
                    
                    if distance < 20 and demand.vulnerability_level == "critical":
                        if local_knowledge > 0.8:
                            evaluation["issues"].append(f"F{facility.id}: Close to vulnerable area D{demand.id}")
                            evaluation["local_score"] -= 0.15
                        else:
                            evaluation["insights"].append(f"F{facility.id}: Community acceptance likely for D{demand.id}")
                            evaluation["local_score"] += 0.1
            
            # Evaluate infrastructure access
            infrastructure_knowledge = expert.domain_knowledge.get("local_infrastructure", 0.8)
            
            for facility in selected_facilities:
                # Simplified infrastructure evaluation
                if facility.x < 20 and facility.y < 30:  # Assume this area has infrastructure challenges
                    if infrastructure_knowledge > 0.7:
                        evaluation["issues"].append(f"F{facility.id}: Limited infrastructure access")
                        evaluation["local_score"] -= 0.1
            
            return evaluation
        
        def _evaluate_financial_implications(self, ai_recommendation: Dict, expert: HumanExpert) -> Dict:
            """Evaluate financial implications"""
            
            evaluation = {"financial_score": 0.0, "concerns": [], "opportunities": []}
            
            total_cost = ai_recommendation["expected_performance"]["total_cost"]
            cost_knowledge = expert.domain_knowledge.get("cost_analysis", 0.8)
            
            # Evaluate cost efficiency
            if total_cost > 1000000:  # High cost
                if cost_knowledge > 0.8:
                    evaluation["concerns"].append(f"High investment cost: ${total_cost:,}")
                    evaluation["financial_score"] -= 0.2
                else:
                    evaluation["opportunities"].append(f"Strategic investment justified by performance")
                    evaluation["financial_score"] += 0.1
            
            # Evaluate ROI potential
            performance = ai_recommendation["expected_performance"]["service_coverage"]
            roi_estimate = (performance * 0.7) / (total_cost / 1000000)  # Simplified ROI
            
            if roi_estimate > 0.15:  # Good ROI
                evaluation["opportunities"].append(f"Strong ROI potential: {roi_estimate:.1%}")
                evaluation["financial_score"] += 0.15
            elif roi_estimate < 0.05:  # Poor ROI
                evaluation["concerns"].append(f"Low ROI potential: {roi_estimate:.1%}")
                evaluation["financial_score"] -= 0.15
            
            return evaluation
        
        def _evaluate_operational_feasibility(self, ai_recommendation: Dict, expert: HumanExpert) -> Dict:
            """Evaluate operational feasibility"""
            
            evaluation = {"operational_score": 0.0, "challenges": [], "strengths": []}
            
            selected_facilities = [f for f in self.facilities if f.id in ai_recommendation["selected_facilities"]]
            ops_knowledge = expert.domain_knowledge.get("operational_efficiency", 0.8)
            
            # Evaluate facility capacity utilization
            total_capacity = sum(f.capacity for f in selected_facilities)
            total_demand = sum(d.base_demand for d in self.demand_zones)
            
            utilization_rate = total_demand / total_capacity
            
            if utilization_rate > 0.9:
                evaluation["challenges"].append(f"High utilization: {utilization_rate:.1%} may limit flexibility")
                evaluation["operational_score"] -= 0.1
            elif utilization_rate < 0.6:
                evaluation["strengths"].append(f"Good capacity buffer: {utilization_rate:.1%} utilization")
                evaluation["operational_score"] += 0.1
            
            # Evaluate geographic distribution
            if len(selected_facilities) >= 2:
                center_x = np.mean([f.x for f in selected_facilities])
                center_y = np.mean([f.y for f in selected_facilities])
                
                max_distance = max([np.sqrt((f.x - center_x)**2 + (f.y - center_y)**2) for f in selected_facilities])
                
                if max_distance > 50:
                    evaluation["challenges"].append(f"Wide geographic distribution may increase transport costs")
                    evaluation["operational_score"] -= 0.1
                elif max_distance < 20:
                    evaluation["strengths"].append(f"Compact geographic distribution reduces transport costs")
                    evaluation["operational_score"] += 0.1
            
            return evaluation
        
        def _evaluate_logistics_optimization(self, ai_recommendation: Dict, expert: HumanExpert) -> Dict:
            """Evaluate logistics optimization aspects"""
            
            evaluation = {"logistics_score": 0.0, "inefficiencies": [], "optimizations": []}
            
            selected_facilities = [f for f in self.facilities if f.id in ai_recommendation["selected_facilities"]]
            logistics_knowledge = expert.domain_knowledge.get("network_design", 0.8)
            
            # Evaluate network connectivity
            connectivity_score = 0
            for demand in self.demand_zones:
                connected_facilities = 0
                for facility in selected_facilities:
                    distance = np.sqrt((facility.x - demand.x)**2 + (facility.y - demand.y)**2)
                    if distance < 40:  # Reasonable service distance
                        connected_facilities += 1
                
                connectivity_score += connected_facilities / len(selected_facilities)
            
            connectivity_score /= len(self.demand_zones)
            
            if connectivity_score > 0.8:
                evaluation["optimizations"].append(f"Excellent network connectivity: {connectivity_score:.1%}")
                evaluation["logistics_score"] += 0.2
            elif connectivity_score < 0.5:
                evaluation["inefficiencies"].append(f"Poor network connectivity: {connectivity_score:.1%}")
                evaluation["logistics_score"] -= 0.2
            
            return evaluation
        
        def _assess_human_confidence(self, ai_recommendation: Dict, expert: HumanExpert) -> float:
            """Assess human confidence in AI recommendation"""
            
            # Base confidence from expert's general confidence
            base_confidence = expert.decision_confidence
            
            # Adjust based on AI confidence
            ai_confidence = ai_recommendation["confidence"]
            confidence_alignment = 1 - abs(base_confidence - ai_confidence)
            
            # Adjust based on uncertainty
            uncertainty = ai_recommendation["uncertainty"]
            uncertainty_penalty = uncertainty * 0.3
            
            # Adjust based on explainability
            explainability_bonus = len(ai_recommendation["reasoning"]) / 10 * 0.2
            
            # Adjust based on alternatives provided
            alternatives_bonus = len(ai_recommendation["alternative_options"]) / 3 * 0.1
            
            confidence = base_confidence * confidence_alignment - uncertainty_penalty + explainability_bonus + alternatives_bonus
            
            return max(0.0, min(1.0, confidence))
        
        def _identify_human_concerns(self, ai_recommendation: Dict, expert: HumanExpert) -> List[str]:
            """Identify human expert concerns"""
            
            concerns = []
            
            # Concern about AI confidence
            if ai_recommendation["confidence"] < 0.7:
                concerns.append("AI confidence appears low, recommendation may be unreliable")
            
            # Concern about uncertainty
            if ai_recommendation["uncertainty"] > 0.3:
                concerns.append("High uncertainty in recommendation, additional analysis needed")
            
            # Concern about risk factors
            if len(ai_recommendation["risk_factors"]) > 2:
                concerns.append("Multiple risk factors identified, comprehensive review required")
            
            # Expert-specific concerns
            if expert.expertise_area == "finance" and ai_recommendation["expected_performance"]["total_cost"] > 800000:
                concerns.append("High investment cost requires additional financial analysis")
            
        
except Exception as e:
    print(f'Error in cell: {e}')
```
