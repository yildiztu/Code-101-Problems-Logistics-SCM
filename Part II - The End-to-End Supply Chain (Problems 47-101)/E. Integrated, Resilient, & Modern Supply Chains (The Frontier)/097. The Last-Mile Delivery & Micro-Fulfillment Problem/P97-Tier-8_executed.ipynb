# 97. The Last-Mile Delivery & Micro-Fulfillment Problem
## Tier 8: The Value-Aligned & Ethical Framework

### Key assumptions
- Delivery algorithms operate within predefined ethical constraints
- Multiple stakeholder values must be balanced (efficiency, fairness, sustainability)
- Constitutional AI principles guide decision-making processes
- Environmental impact limits and social equity requirements are enforced
- Worker welfare standards and customer privacy protection are prioritized

### Approach (step-by-step)
1. **Ethical Framework Definition**: Establish constitutional AI principles and constraints
2. **Stakeholder Identification**: Define all affected parties and their interests
3. **Value Integration**: Incorporate ethical constraints into optimization algorithms
4. **Multi-Objective Optimization**: Balance competing values with explicit trade-offs
5. **Transparency & Accountability**: Ensure decision rationale is publicly documented

### What to look for in the results
- Ethical constraint compliance and violation detection
- Stakeholder utility distribution and fairness metrics
- Environmental impact measurements and carbon footprint analysis
- Trade-off resolution between efficiency and ethical considerations

### Concrete example (from the source)
System faces holiday delivery dilemma: most efficient route concentrates 67% of traffic in affluent areas. Ethical framework triggers constraint violation and re-optimizes to ensure equitable service distribution while accepting 7% efficiency penalty.

```python
from typing import Tuple, List, Dict, Set

# Import required libraries for Ethical Framework
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass
from typing import List, Tuple, Dict, Optional
import random
import time
from datetime import datetime, timedelta
from enum import Enum
import warnings
warnings.filterwarnings('ignore')

# Set style for visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
class EthicalPrinciple(Enum):
    ENVIRONMENTAL_SUSTAINABILITY = "environmental_sustainability"
    SOCIAL_EQUITY = "social_equity"
    WORKER_WELFARE = "worker_welfare"
    CUSTOMER_PRIVACY = "customer_privacy"
    FAIR_COMPETITION = "fair_competition"
    TRANSPARENCY = "transparency"

@dataclass
class EthicalConstraint:
    """Represents an ethical constraint in the system"""
    principle: EthicalPrinciple
    constraint_type: str  # 'hard', 'soft', 'guideline'
    description: str
    metric_name: str
    threshold_value: float
    penalty_weight: float  # Weight in objective function
    measurement_method: str

@dataclass
class Stakeholder:
    """Represents a stakeholder in the delivery system"""
    id: int
    name: str
    type: str  # 'customer', 'worker', 'community', 'environment', 'company'
    interests: List[str]
    utility_function: Dict[str, float]  # weights for different outcomes
    priority: float  # 0-1 scale
    location: Optional[Tuple[float, float]] = None
    demographic_data: Optional[Dict] = None

@dataclass
class DeliveryDecision:
    """Represents a delivery decision with ethical implications"""
    id: int
    customer_id: int
    origin: Tuple[float, float]
    destination: Tuple[float, float]
    priority: int  # 1-3 scale
    time_window: Tuple[datetime, datetime]
    package_size: float
    neighborhood_type: str  # 'affluent', 'middle', 'low_income'
    environmental_sensitivity: float  # 0-1 scale
    worker_safety_risk: float  # 0-1 scale

@dataclass
class EthicalAssessment:
    """Assessment of decision's ethical compliance"""
    decision_id: int
    constraint_compliance: Dict[EthicalPrinciple, bool]
    constraint_violations: List[EthicalPrinciple]
    stakeholder_impact: Dict[int, float]
    ethical_score: float  # 0-1 scale
    trade_offs_identified: List[str]
    recommendation: str

class ConstitutionalAI:
    """Constitutional AI system with ethical framework"""
    
    def __init__(self):
        self.constitution = self._establish_constitution()
        self.ethical_constraints = self._define_ethical_constraints()
        self.stakeholders = []
        self.decision_history = []
        self.impact_metrics = {
            'environmental_impact': 0.0,
            'social_equity_score': 0.0,
            'worker_welfare_index': 0.0,
            'privacy_protection_score': 0.0
        }
        
    def _establish_constitution(self) -> Dict[str, str]:
        """Establish constitutional principles for the AI system"""
        return {
            'principle_1': 'All delivery decisions must minimize environmental impact',
            'principle_2': 'Service must be equitably distributed across all socioeconomic areas',
            'principle_3': 'Worker safety and welfare shall not be compromised for efficiency',
            'principle_4': 'Customer privacy and data protection are fundamental rights',
            'principle_5': 'Transparency in decision-making is required for accountability',
            'principle_6': 'Fair competition and market integrity must be maintained'
        }
    
    def _define_ethical_constraints(self) -> List[EthicalConstraint]:
        """Define specific ethical constraints"""
        constraints = [
            # Environmental constraints
            EthicalConstraint(
                principle=EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY,
                constraint_type='hard',
                description='Carbon emissions per delivery must not exceed baseline by 40%',
                metric_name='carbon_emissions',
                threshold_value=1.4,  # 140% of baseline
                penalty_weight=0.3,
                measurement_method='emissions_tracking'
            ),
            EthicalConstraint(
                principle=EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY,
                constraint_type='soft',
                description='Prefer low-emission delivery methods when feasible',
                metric_name='emission_preference',
                threshold_value=0.7,
                penalty_weight=0.1,
                measurement_method='mode_selection'
            ),
            
            # Social equity constraints
            EthicalConstraint(
                principle=EthicalPrinciple.SOCIAL_EQUITY,
                constraint_type='hard',
                description='Service quality difference between neighborhoods must not exceed 15%',
                metric_name='service_equity',
                threshold_value=0.15,
                penalty_weight=0.4,
                measurement_method='quality_comparison'
            ),
            EthicalConstraint(
                principle=EthicalPrinciple.SOCIAL_EQUITY,
                constraint_type='soft',
                description='Maintain minimum service levels in underserved areas',
                metric_name='minimum_service',
                threshold_value=0.8,
                penalty_weight=0.2,
                measurement_method='service_level_monitoring'
            ),
            
            # Worker welfare constraints
            EthicalConstraint(
                principle=EthicalPrinciple.WORKER_WELFARE,
                constraint_type='hard',
                description='Maximum driving hours shall not exceed 10 hours per shift',
                metric_name='driving_hours',
                threshold_value=10.0,
                penalty_weight=0.5,
                measurement_method='time_tracking'
            ),
            EthicalConstraint(
                principle=EthicalPrinciple.WORKER_WELFARE,
                constraint_type='soft',
                description='Provide adequate rest breaks between deliveries',
                metric_name='break_compliance',
                threshold_value=0.9,
                penalty_weight=0.1,
                measurement_method='schedule_monitoring'
            ),
            
            # Privacy constraints
            EthicalConstraint(
                principle=EthicalPrinciple.CUSTOMER_PRIVACY,
                constraint_type='hard',
                description='Customer location data minimization and consent-based tracking',
                metric_name='data_minimization',
                threshold_value=0.3,
                penalty_weight=0.3,
                measurement_method='privacy_audit'
            ),
        ]
        
        return constraints
    
    def add_stakeholder(self, stakeholder: Stakeholder):
        """Add stakeholder to the system"""
        self.stakeholders.append(stakeholder)
    
    def assess_decision_ethics(self, decision: DeliveryDecision, 
                               proposed_solution: Dict) -> EthicalAssessment:
        """Assess ethical compliance of a delivery decision"""
        
        # Initialize compliance tracking
        compliance = {}
        violations = []
        stakeholder_impacts = {}
        trade_offs = []
        
        # Check each constraint
        for constraint in self.ethical_constraints:
            compliance_result = self._check_constraint_compliance(decision, proposed_solution, constraint)
            compliance[constraint.principle] = compliance_result['complies']
            
            if not compliance_result['complies']:
                violations.append(constraint.principle)
                trade_offs.append(f"{constraint.description}: {compliance_result['violation_details']}")
            
            # Calculate stakeholder impacts
            self._calculate_stakeholder_impacts(decision, proposed_solution, constraint, stakeholder_impacts)
        
        # Calculate overall ethical score
        ethical_score = self._calculate_ethical_score(compliance, violations)
        
        # Generate recommendation
        recommendation = self._generate_ethical_recommendation(compliance, violations, ethical_score)
        
        return EthicalAssessment(
            decision_id=decision.id,
            constraint_compliance=compliance,
            constraint_violations=violations,
            stakeholder_impact=stakeholder_impacts,
            ethical_score=ethical_score,
            trade_offs_identified=trade_offs,
            recommendation=recommendation
        )
    
    def _check_constraint_compliance(self, decision: DeliveryDecision, 
                                   proposed_solution: Dict, constraint: EthicalConstraint) -> Dict:
        """Check if a decision complies with a specific constraint"""
        
        if constraint.principle == EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY:
            if constraint.metric_name == 'carbon_emissions':
                # Calculate emissions for proposed solution
                distance = proposed_solution.get('distance', 10.0)  # km
                vehicle_type = proposed_solution.get('vehicle_type', 'van')
                
                # Emission factors (kg CO2 per km)
                emission_factors = {
                    'van': 0.25,
                    'bike': 0.0,
                    'drone': 0.15,
                    'robot': 0.05
                }
                
                baseline_emissions = 10.0 * 0.25  # 10km van baseline
                actual_emissions = distance * emission_factors.get(vehicle_type, 0.25)
                emissions_ratio = actual_emissions / baseline_emissions
                
                complies = emissions_ratio <= constraint.threshold_value
                violation_details = f"Emissions ratio: {emissions_ratio:.2f} > {constraint.threshold_value:.2f}" if not complies else ""
                
                return {
                    'complies': complies,
                    'measured_value': emissions_ratio,
                    'threshold': constraint.threshold_value,
                    'violation_details': violation_details
                }
        
        elif constraint.principle == EthicalPrinciple.SOCIAL_EQUITY:
            if constraint.metric_name == 'service_equity':
                # Check service distribution across neighborhoods
                neighborhood_service = proposed_solution.get('neighborhood_service', {})
                
                # Calculate service quality differences
                service_levels = list(neighborhood_service.values())
                if len(service_levels) >= 2:
                    max_service = max(service_levels)
                    min_service = min(service_levels)
                    service_difference = (max_service - min_service) / max_service
                    
                    complies = service_difference <= constraint.threshold_value
                    violation_details = f"Service difference: {service_difference:.2f} > {constraint.threshold_value:.2f}" if not complies else ""
                    
                    return {
                        'complies': complies,
                        'measured_value': service_difference,
                        'threshold': constraint.threshold_value,
                        'violation_details': violation_details
                    }
        
        elif constraint.principle == EthicalPrinciple.WORKER_WELFARE:
            if constraint.metric_name == 'driving_hours':
                # Check worker hours
                worker_hours = proposed_solution.get('worker_hours', 8.0)
                
                complies = worker_hours <= constraint.threshold_value
                violation_details = f"Worker hours: {worker_hours:.1f} > {constraint.threshold_value:.1f}" if not complies else ""
                
                return {
                    'complies': complies,
                    'measured_value': worker_hours,
                    'threshold': constraint.threshold_value,
                    'violation_details': violation_details
                }
        
        elif constraint.principle == EthicalPrinciple.CUSTOMER_PRIVACY:
            if constraint.metric_name == 'data_minimization':
                # Check data collection practices
                data_collected = proposed_solution.get('data_collected', ['location', 'time', 'signature'])
                data_score = len(data_collected) / 10.0  # Normalize to 0-1
                
                complies = data_score <= constraint.threshold_value
                violation_details = f"Data collection score: {data_score:.2f} > {constraint.threshold_value:.2f}" if not complies else ""
                
                return {
                    'complies': complies,
                    'measured_value': data_score,
                    'threshold': constraint.threshold_value,
                    'violation_details': violation_details
                }
        
        # Default compliance (if constraint type not specifically handled)
        return {
            'complies': True,
            'measured_value': 0.0,
            'threshold': constraint.threshold_value,
            'violation_details': ''
        }
    
    def _calculate_stakeholder_impacts(self, decision: DeliveryDecision, 
                                   proposed_solution: Dict, constraint: EthicalConstraint, 
                                   stakeholder_impacts: Dict):
        """Calculate impacts on stakeholders"""
        for stakeholder in self.stakeholders:
            impact = 0.0
            
            if stakeholder.type == 'customer':
                # Customer impact based on service quality and privacy
                service_quality = proposed_solution.get('service_quality', 0.8)
                privacy_impact = -0.1 if constraint.principle == EthicalPrinciple.CUSTOMER_PRIVACY else 0.0
                impact = service_quality + privacy_impact
            
            elif stakeholder.type == 'worker':
                # Worker impact based on welfare and safety
                safety_score = 1.0 - proposed_solution.get('worker_safety_risk', 0.1)
                hours_impact = -0.1 if proposed_solution.get('worker_hours', 8.0) > 9 else 0.0
                impact = safety_score + hours_impact
            
            elif stakeholder.type == 'community':
                # Community impact based on environmental and equity factors
                env_impact = -0.1 if constraint.principle == EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY else 0.0
                equity_impact = 0.1 if constraint.principle == EthicalPrinciple.SOCIAL_EQUITY else 0.0
                impact = env_impact + equity_impact
            
            elif stakeholder.type == 'environment':
                # Environmental impact (negative is bad for environment)
                impact = -0.2 if constraint.principle == EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY else 0.0
            
            elif stakeholder.type == 'company':
                # Company impact based on efficiency and reputation
                efficiency = proposed_solution.get('efficiency', 0.8)
                reputation_impact = 0.1 if len([c for c in self.ethical_constraints if not compliance.get(c.principle, True)]) == 0 else -0.2
                impact = efficiency + reputation_impact
            
            stakeholder_impacts[stakeholder.id] = impact
    
    def _calculate_ethical_score(self, compliance: Dict[EthicalPrinciple, bool], 
                                violations: List[EthicalPrinciple]) -> float:
        """Calculate overall ethical score"""
        total_constraints = len(self.ethical_constraints)
        complied_constraints = sum(1 for complies in compliance.values() if complies)
        
        # Base score from compliance
        base_score = complied_constraints / total_constraints
        
        # Penalty for hard constraint violations
        hard_violations = len([c for c in violations if any(constraint.principle == c and constraint.constraint_type == 'hard' for constraint in self.ethical_constraints)])
        penalty = hard_violations * 0.2
        
        return max(0.0, base_score - penalty)
    
    def _generate_ethical_recommendation(self, compliance: Dict[EthicalPrinciple, bool], 
                                        violations: List[EthicalPrinciple], ethical_score: float) -> str:
        """Generate recommendation based on ethical assessment"""
        
        if ethical_score >= 0.9:
            return "Decision fully complies with ethical framework. Proceed as recommended."
        elif ethical_score >= 0.7:
            return "Decision mostly complies with ethical framework. Minor adjustments recommended."
        elif ethical_score >= 0.5:
            return "Decision has significant ethical concerns. Re-optimization required."
        else:
            return "Decision violates core ethical principles. Major revision required or rejection."

class EthicalOptimizer:
    """Multi-objective optimizer with ethical constraints"""
    
    def __init__(self, constitutional_ai: ConstitutionalAI):
        self.ai = constitutional_ai
        self.optimization_history = []
        
    def optimize_with_ethics(self, decisions: List[DeliveryDecision]) -> Dict:
        """Optimize delivery decisions with ethical constraints"""
        
        print("Running ethical optimization...")
        
        optimization_results = {
            'original_solution': self._generate_baseline_solution(decisions),
            'ethical_assessments': [],
            'ethical_optimization': None,
            'trade_off_analysis': None,
            'final_solution': None
        }
        
        # Assess original solution
        print("\n📋 Assessing baseline solution...")
        original_assessments = []
        for decision in decisions:
            assessment = self.ai.assess_decision_ethics(decision, optimization_results['original_solution'])
            original_assessments.append(assessment)
            optimization_results['ethical_assessments'].append(assessment)
        
        # Check for violations
        violations = self._identify_violations(original_assessments)
        
        if violations:
            print(f"⚠️ Found {len(violations)} ethical violations in baseline solution")
            
            # Run ethical re-optimization
            print("\n🔄 Running ethical re-optimization...")
            ethical_solution = self._ethical_reoptimization(decisions, violations)
            optimization_results['ethical_optimization'] = ethical_solution
            
            # Assess ethical solution
            ethical_assessments = []
            for decision in decisions:
                assessment = self.ai.assess_decision_ethics(decision, ethical_solution)
                ethical_assessments.append(assessment)
            
            optimization_results['final_solution'] = ethical_solution
            optimization_results['ethical_assessments'] = ethical_assessments
            
            # Analyze trade-offs
            trade_off_analysis = self._analyze_trade_offs(
                optimization_results['original_solution'],
                ethical_solution,
                original_assessments,
                ethical_assessments
            )
            optimization_results['trade_off_analysis'] = trade_off_analysis
            
        else:
            print("✅ Baseline solution complies with all ethical constraints")
            optimization_results['final_solution'] = optimization_results['original_solution']
        
        return optimization_results
    
    def _generate_baseline_solution(self, decisions: List[DeliveryDecision]) -> Dict:
        """Generate baseline solution (efficiency-focused)"""
        
        # Simulate efficiency-optimized solution
        solution = {
            'total_distance': 0.0,
            'total_time': 0.0,
            'total_cost': 0.0,
            'vehicle_assignments': [],
            'neighborhood_service': {},
            'efficiency': 0.95,  # High efficiency
            'carbon_emissions': 1.6,  # Above threshold
            'service_equity': 0.25,  # Above threshold (poor equity)
            'worker_hours': 11.2,  # Above threshold
            'data_collected': ['location', 'time', 'signature', 'photo', 'customer_id', 'package_contents'],
            'worker_safety_risk': 0.3
        }
        
        # Calculate metrics based on decisions
        neighborhood_stats = {'affluent': 0, 'middle': 0, 'low_income': 0}
        neighborhood_quality = {'affluent': 0.95, 'middle': 0.85, 'low_income': 0.70}
        
        for decision in decisions:
            # Simulate distance calculation
            distance = np.sqrt((decision.destination[0] - decision.origin[0])**2 + 
                           (decision.destination[1] - decision.origin[1])**2)
            solution['total_distance'] += distance
            
            # Track neighborhood distribution
            neighborhood_stats[decision.neighborhood_type] += 1
        
        # Calculate neighborhood service levels
        for neighborhood, count in neighborhood_stats.items():
            if count > 0:
                solution['neighborhood_service'][neighborhood] = neighborhood_quality[neighborhood]
        
        return solution
    
    def _identify_violations(self, assessments: List[EthicalAssessment]) -> List[EthicalPrinciple]:
        """Identify ethical violations from assessments"""
        violations = set()
        
        for assessment in assessments:
            violations.update(assessment.constraint_violations)
        
        return list(violations)
    
    def _ethical_reoptimization(self, decisions: List[DeliveryDecision], 
                               violations: List[EthicalPrinciple]) -> Dict:
        """Re-optimize solution to address ethical violations"""
        
        # Start with baseline and modify to address violations
        ethical_solution = self._generate_baseline_solution(decisions)
        
        # Adjust solution based on violations
        if EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY in violations:
            # Reduce emissions by using greener vehicles and optimizing routes
            ethical_solution['carbon_emissions'] = 1.2  # Below threshold
            ethical_solution['efficiency'] -= 0.05  # Small efficiency penalty
        
        if EthicalPrinciple.SOCIAL_EQUITY in violations:
            # Improve service equity by redistributing resources
            ethical_solution['service_equity'] = 0.12  # Below threshold
            ethical_solution['neighborhood_service']['low_income'] = 0.88
            ethical_solution['neighborhood_service']['affluent'] = 0.85
            ethical_solution['efficiency'] -= 0.07  # Efficiency penalty for equity
        
        if EthicalPrinciple.WORKER_WELFARE in violations:
            # Reduce worker hours
            ethical_solution['worker_hours'] = 9.1  # Below threshold
            ethical_solution['efficiency'] -= 0.03  # Small penalty
        
        if EthicalPrinciple.CUSTOMER_PRIVACY in violations:
            # Reduce data collection
            ethical_solution['data_collected'] = ['location', 'time']  # Minimal data
            ethical_solution['efficiency'] -= 0.02  # Small penalty
        
        return ethical_solution
    
    def _analyze_trade_offs(self, original_solution: Dict, ethical_solution: Dict,
                          original_assessments: List[EthicalAssessment], 
                          ethical_assessments: List[EthicalAssessment]) -> Dict:
        """Analyze trade-offs between efficiency and ethics"""
        
        efficiency_loss = original_solution['efficiency'] - ethical_solution['efficiency']
        cost_increase = efficiency_loss * 1000  # Assume $1000 per efficiency point
        
        original_score = np.mean([a.ethical_score for a in original_assessments])
        ethical_score = np.mean([a.ethical_score for a in ethical_assessments])
        ethical_improvement = ethical_score - original_score
        
        return {
            'efficiency_loss': efficiency_loss,
            'cost_increase': cost_increase,
            'ethical_improvement': ethical_improvement,
            'violations_resolved': len([a for a in original_assessments if a.constraint_violations]) - len([a for a in ethical_assessments if a.constraint_violations]),
            'recommendation': f"Accept {efficiency_loss*100:.1f}% efficiency loss for {ethical_improvement*100:.1f}% ethical improvement"
        }

print("Ethical Framework implementation complete")
```
