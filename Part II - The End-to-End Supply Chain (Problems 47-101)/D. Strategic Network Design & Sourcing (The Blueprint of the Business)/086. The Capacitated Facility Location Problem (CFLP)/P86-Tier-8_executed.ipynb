# 86. The Capacitated Facility Location Problem (CFLP)

## Tier 8: The Value-Aligned & Ethical Framework (Fairness Constraints)

### Key Assumptions
- Cost optimization alone may create societal inequities and externalities
- Ethical principles must be embedded in optimization models
- Multiple stakeholders have legitimate claims on facility location decisions
- Fairness can be quantified and optimized alongside efficiency
- Long-term societal value may differ from short-term cost minimization

### Approach (Step-by-Step)

The Value-Aligned & Ethical Framework transforms CFLP from a pure cost minimization problem to a multi-objective optimization that balances efficiency with equity, sustainability, and social responsibility.

**Ethical Optimization Framework:**
1. **Stakeholder Identification**: Map all affected parties and their interests
2. **Ethical Principle Definition**: Establish fairness, equity, and sustainability metrics
3. **Constraint Formulation**: Convert ethical principles into mathematical constraints
4. **Multi-Objective Optimization**: Balance cost with ethical objectives
5. **Pareto Front Analysis**: Generate trade-off curves for informed decision making
6. **Stakeholder Evaluation**: Involve stakeholders in final solution selection

### What to Look for in the Results
- How ethical constraints affect facility location patterns
- Trade-offs between cost efficiency and equity metrics
- Pareto front showing optimal balance points
- Stakeholder satisfaction with different solution strategies

### Concrete Example

We'll demonstrate ethical CFLP optimization:
- **8 potential facilities** with different community impacts
- **10 customer zones** with varying socioeconomic characteristics
- **5 stakeholder groups**: Communities, Employees, Environment, Investors, Regulators
- **3 ethical objectives**: Economic efficiency, Social equity, Environmental sustainability
- **Pareto optimization** generating optimal trade-off solutions

This shows how ethical frameworks can transform optimization from pure efficiency to value-aligned decision making.

```python
@dataclass
class Facility:
    """Facility with ethical and social impact attributes"""
    id: int
    name: str
    x_coord: float
    y_coord: float
    fixed_cost: float
    capacity: float
    # Ethical attributes
    community_impact: float      # 0-1, impact on local community
    job_creation: int           # Number of jobs created
    local_tax_revenue: float     # Annual tax revenue for community
    environmental_footprint: float # 0-1, environmental impact (lower is better)
    accessibility_score: float   # 0-1, accessibility for disabled/elderly
    safety_rating: float         # 0-1, workplace safety
    
@dataclass
class CustomerZone:
    """Customer zone with socioeconomic characteristics"""
    id: int
    name: str
    x_coord: float
    y_coord: float
    demand: float
    # Socioeconomic attributes
    median_income: float      # Median household income
    unemployment_rate: float   # Unemployment rate (0-1)
    population_density: float   # Population per sq km
    age_demographics: Dict[str, float]  # Age group percentages
    service_desert: float      # Service desert indicator (0-1)
    strategic_importance: float # 0-1, business importance
    
@dataclass
class Stakeholder:
    """Stakeholder with preferences and weights"""
    id: int
    name: str
    type: str  # COMMUNITY, EMPLOYEE, ENVIRONMENT, INVESTOR, REGULATOR
    objectives: List[str]     # Key objectives
    weights: Dict[str, float] # Weights for different objectives
    veto_power: float         # 0-1, ability to veto solutions
    
@dataclass
class EthicalMetrics:
    """Ethical performance metrics"""
    economic_efficiency: float   # Cost efficiency score
    social_equity: float         # Equity and fairness score
    environmental_sustainability: float  # Environmental impact score
    stakeholder_satisfaction: float # Overall stakeholder satisfaction
    long_term_value: float       # Long-term societal value
    
@dataclass
class EthicalSolution:
    """Solution with ethical evaluation"""
    open_facilities: Set[int]
    total_cost: float
    service_level: float
    ethical_metrics: EthicalMetrics
    stakeholder_evaluations: Dict[int, float]
    pareto_dominated: bool
```

```python
class EthicalCFLOptimizer:
    """Ethical CFLP optimizer with multi-objective capabilities"""
    
    def __init__(self, facilities: List[Facility], 
                 customer_zones: List[CustomerZone],
                 stakeholders: List[Stakeholder]):
        self.facilities = facilities
        self.customer_zones = customer_zones
        self.stakeholders = stakeholders
        self.transport_costs = self._calculate_transport_costs()
        
    def _calculate_transport_costs(self) -> Dict[Tuple[int, int], float]:
        """Calculate transportation costs"""
        costs = {}
        for facility in self.facilities:
            for zone in self.customer_zones:
                distance = np.sqrt((facility.x_coord - zone.x_coord)**2 + 
                                  (facility.y_coord - zone.y_coord)**2)
                costs[(facility.id, zone.id)] = distance * 10.0  # $10 per unit per distance
        return costs
    
    def evaluate_basic_solution(self, open_facilities: Set[int]) -> Dict:
        """Evaluate basic solution without ethical considerations"""
        
        # Calculate fixed costs
        fixed_cost = sum(f.fixed_cost for f in self.facilities if f.id in open_facilities)
        
        # Greedy customer assignment
        transport_cost = 0
        facility_loads = {f.id: 0 for f in self.facilities if f.id in open_facilities}
        assignments = {}
        unmet_demand = 0
        
        sorted_zones = sorted(self.customer_zones, key=lambda z: z.demand, reverse=True)
        
        for zone in sorted_zones:
            best_facility = None
            best_cost = float('inf')
            
            for facility_id in open_facilities:
                facility = next(f for f in self.facilities if f.id == facility_id)
                
                if facility_loads[facility_id] + zone.demand <= facility.capacity:
                    cost = self.transport_costs[(facility_id, zone.id)]
                    if cost < best_cost:
                        best_cost = cost
                        best_facility = facility_id
            
            if best_facility:
                assignments[zone.id] = best_facility
                facility_loads[best_facility] += zone.demand
                transport_cost += best_cost * zone.demand
            else:
                unmet_demand += zone.demand
        
        total_cost = fixed_cost + transport_cost
        total_demand = sum(z.demand for z in self.customer_zones)
        service_level = (total_demand - unmet_demand) / total_demand if total_demand > 0 else 0
        
        return {
            'total_cost': total_cost,
            'fixed_cost': fixed_cost,
            'transport_cost': transport_cost,
            'service_level': service_level,
            'assignments': assignments,
            'facility_loads': facility_loads,
            'unmet_demand': unmet_demand
        }
    
    def calculate_ethical_metrics(self, open_facilities: Set[int]) -> EthicalMetrics:
        """Calculate comprehensive ethical metrics"""
        
        basic_solution = self.evaluate_basic_solution(open_facilities)
        
        # Economic efficiency (normalized)
        max_cost = sum(f.fixed_cost for f in self.facilities)
        min_cost = min(f.fixed_cost for f in self.facilities)
        cost_efficiency = 1 - (basic_solution['total_cost'] - min_cost) / (max_cost - min_cost)
        
        # Social equity calculations
        equity_metrics = self._calculate_social_equity(open_facilities, basic_solution['assignments'])
        
        # Environmental sustainability
        environmental_metrics = self._calculate_environmental_impact(open_facilities)
        
        # Stakeholder satisfaction
        stakeholder_scores = self._calculate_stakeholder_satisfaction(
            open_facilities, basic_solution, equity_metrics, environmental_metrics
        )
        
        # Long-term value (simplified)
        long_term_value = (
            0.3 * cost_efficiency +
            0.3 * equity_metrics['service_equity'] +
            0.2 * environmental_metrics['sustainability_score'] +
            0.2 * stakeholder_scores
        )
        
        return EthicalMetrics(
            economic_efficiency=cost_efficiency,
            social_equity=equity_metrics['overall_equity'],
            environmental_sustainability=environmental_metrics['sustainability_score'],
            stakeholder_satisfaction=stakeholder_scores,
            long_term_value=long_term_value
        )
    
    def _calculate_social_equity(self, open_facilities: Set[int], 
                             assignments: Dict[int, int]) -> Dict:
        """Calculate social equity metrics"""
        
        # Service equity: equal access across different socioeconomic groups
        zone_service_levels = {}
        for zone in self.customer_zones:
            if zone.id in assignments:
                facility_id = assignments[zone.id]
                facility = next(f for f in self.facilities if f.id == facility_id)
                # Service quality based on facility attributes
                service_quality = (facility.accessibility_score + facility.safety_rating) / 2
                zone_service_levels[zone.id] = service_quality
            else:
                zone_service_levels[zone.id] = 0
        
        # Calculate equity metrics
        service_level_variance = np.var(list(zone_service_levels.values()))
        service_level_mean = np.mean(list(zone_service_levels.values()))
        service_equity = 1 - (service_level_variance / (service_level_mean + 0.01))  # Avoid division by zero
        
        # Geographic equity: distribution across different areas
        facility_coverage = {fid: 0 for fid in open_facilities}
        for zone_id, facility_id in assignments.items():
            facility_coverage[facility_id] += 1
        
        # Calculate coverage balance
        if facility_coverage:
            max_coverage = max(facility_coverage.values())
            coverage_balance = 1 - (np.std(list(facility_coverage.values())) / (max_coverage + 0.01))
        else:
            coverage_balance = 0
        
        # Economic equity: job creation and tax revenue distribution
        total_jobs = sum(f.job_creation for f in self.facilities if f.id in open_facilities)
        total_tax_revenue = sum(f.local_tax_revenue for f in self.facilities if f.id in open_facilities)
        
        # Calculate distribution fairness
        if len(open_facilities) > 0:
            avg_jobs = total_jobs / len(open_facilities)
            avg_tax = total_tax_revenue / len(open_facilities)
            
            job_distribution = 1 - (np.std([f.job_creation for f in self.facilities if f.id in open_facilities]) / (avg_jobs + 1))
            tax_distribution = 1 - (np.std([f.local_tax_revenue for f in self.facilities if f.id in open_facilities]) / (avg_tax + 1))
        else:
            job_distribution = 0
            tax_distribution = 0
        
        return {
            'service_equity': service_equity,
            'coverage_balance': coverage_balance,
            'job_distribution': job_distribution,
            'tax_distribution': tax_distribution,
            'overall_equity': (service_equity + coverage_balance + job_distribution + tax_distribution) / 4
        }
    
    def _calculate_environmental_impact(self, open_facilities: Set[int]) -> Dict:
        """Calculate environmental impact metrics"""
        
        # Average environmental footprint
        if open_facilities:
            avg_footprint = np.mean([f.environmental_footprint for f in self.facilities if f.id in open_facilities])
        else:
            avg_footprint = 0.5  # Default
        
        # Sustainability score (lower footprint is better)
        sustainability_score = 1 - avg_footprint
        
        # Community impact balance
        if open_facilities:
            avg_community_impact = np.mean([f.community_impact for f in self.facilities if f.id in open_facilities])
        else:
            avg_community_impact = 0.5
        
        # Environmental-community balance
        env_community_balance = 1 - abs(avg_footprint - avg_community_impact)
        
        return {
            'sustainability_score': sustainability_score,
            'env_community_balance': env_community_balance,
            'avg_footprint': avg_footprint,
            'avg_community_impact': avg_community_impact
        }
    
    def _calculate_stakeholder_satisfaction(self, open_facilities: Set[int],
                                   basic_solution: Dict,
                                   equity_metrics: Dict,
                                   environmental_metrics: Dict) -> float:
        """Calculate stakeholder satisfaction"""
        
        total_satisfaction = 0
        total_weight = 0
        
        for stakeholder in self.stakeholders:
            satisfaction = 0
            weight_sum = sum(stakeholder.weights.values())
            
            # Evaluate based on stakeholder objectives
            for objective, weight in stakeholder.weights.items():
                if objective == "total_cost":
                    # Lower cost is better for investors
                    max_cost = sum(f.fixed_cost for f in self.facilities)
                    cost_score = 1 - (basic_solution['total_cost'] / max_cost)
                    satisfaction += weight * cost_score
                elif objective == "service_level":
                    satisfaction += weight * basic_solution['service_level']
                elif objective in equity_metrics:
                    satisfaction += weight * equity_metrics[objective]
                elif objective in environmental_metrics:
                    satisfaction += weight * environmental_metrics[objective]
                elif objective == "job_creation":
                    total_jobs = sum(f.job_creation for f in self.facilities if f.id in open_facilities)
                    max_jobs = sum(f.job_creation for f in self.facilities)
                    job_score = total_jobs / max_jobs if max_jobs > 0 else 0
                    satisfaction += weight * job_score
                elif objective == "local_tax_revenue":
                    total_tax = sum(f.local_tax_revenue for f in self.facilities if f.id in open_facilities)
                    max_tax = sum(f.local_tax_revenue for f in self.facilities)
                    tax_score = total_tax / max_tax if max_tax > 0 else 0
                    satisfaction += weight * tax_score
            
            total_satisfaction += stakeholder.veto_power * satisfaction
            total_weight += stakeholder.veto_power
        
        if total_weight > 0:
            return total_satisfaction / total_weight
        else:
            return 0
    
    def generate_pareto_solutions(self, n_solutions: int = 50) -> List[EthicalSolution]:
        """Generate Pareto-optimal solutions"""
        
        solutions = []
        
        # Generate diverse solutions using different strategies
        strategies = [
            self._generate_cost_optimized_solution,
            self._generate_equity_focused_solution,
            self._generate_environmental_solution,
            self._generate_balanced_solution
        ]
        
        for strategy in strategies:
            for i in range(n_solutions // len(strategies)):
                try:
                    solution = strategy()
                    if solution and not solution.pareto_dominated:
                        solutions.append(solution)
                except:
                    continue
        
        # Check for Pareto dominance
        for i, solution1 in enumerate(solutions):
            for j, solution2 in enumerate(solutions):
                if i != j and self._dominates(solution2, solution1):
                    solution1.pareto_dominated = True
                elif i != j and self._dominates(solution1, solution2):
                    solution2.pareto_dominated = True
        
        # Remove dominated solutions
        pareto_solutions = [s for s in solutions if not s.pareto_dominated]
        
        return pareto_solutions
    
    def _dominates(self, solution1: EthicalSolution, solution2: EthicalSolution) -> bool:
        """Check if solution1 dominates solution2"""
        
        # Solution1 dominates solution2 if it's better or equal in all objectives
        # and strictly better in at least one objective
        better_or_equal = (
            solution1.ethical_metrics.economic_efficiency >= solution2.ethical_metrics.economic_efficiency and
            solution1.ethical_metrics.social_equity >= solution2.ethical_metrics.social_equity and
            solution1.ethical_metrics.environmental_sustainability >= solution2.ethical_metrics.environmental_sustainability and
            solution1.ethical_metrics.stakeholder_satisfaction >= solution2.ethical_metrics.stakeholder_satisfaction
        )
        
        strictly_better = (
            solution1.ethical_metrics.economic_efficiency > solution2.ethical_metrics.economic_efficiency or
            solution1.ethical_metrics.social_equity > solution2.ethical_metrics.social_equity or
            solution1.ethical_metrics.environmental_sustainability > solution2.ethical_metrics.environmental_sustainability or
            solution1.ethical_metrics.stakeholder_satisfaction > solution2.ethical_metrics.stakeholder_satisfaction
        )
        
        return better_or_equal and strictly_better
    
    def _generate_cost_optimized_solution(self) -> Optional[EthicalSolution]:
        """Generate cost-optimized solution"""
        
        # Sort facilities by cost (lowest first)
        facility_costs = [(f.id, f.fixed_cost) for f in self.facilities]
        facility_costs.sort(key=lambda x: x[1])
        
        open_facilities = set()
        evaluation = self.evaluate_basic_solution(open_facilities)
        
        # Add facilities until service level is acceptable
        for facility_id, _ in facility_costs:
            if evaluation['service_level'] < 0.9:
                open_facilities.add(facility_id)
                evaluation = self.evaluate_basic_solution(open_facilities)
            else:
                break
        
        if open_facilities:
            ethical_metrics = self.calculate_ethical_metrics(open_facilities)
            
            return EthicalSolution(
                open_facilities=open_facilities,
                total_cost=evaluation['total_cost'],
                service_level=evaluation['service_level'],
                ethical_metrics=ethical_metrics,
                stakeholder_evaluations={},
                pareto_dominated=False
            )
        else:
            return None
    
    def _generate_equity_focused_solution(self) -> Optional[EthicalSolution]:
        """Generate equity-focused solution"""
        
        # Prioritize facilities that serve underserved areas
        zone_priorities = {}
        for zone in self.customer_zones:
            # Higher priority for service deserts and low-income areas
            priority = (
                0.4 * zone.service_desert +
                0.3 * (1 - zone.median_income / 100000) +
                0.3 * zone.strategic_importance
            )
            zone_priorities[zone.id] = priority
        
        # Sort zones by priority (highest first)
        sorted_zones = sorted(self.customer_zones, key=lambda z: zone_priorities[z.id], reverse=True)
        
        open_facilities = set()
        evaluation = self.evaluate_basic_solution(open_facilities)
        
        # Add facilities to serve high-priority zones
        for zone in sorted_zones:
            # Find best facility for this zone
            best_facility = None
            best_cost = float('inf')
            
            for facility in self.facilities:
                if facility.id not in open_facilities:
                    cost = self.transport_costs[(facility.id, zone.id)]
                    if cost < best_cost:
                        best_cost = cost
                        best_facility = facility.id
            
            if best_facility:
                open_facilities.add(best_facility)
                evaluation = self.evaluate_basic_solution(open_facilities)
            
        if open_facilities:
            ethical_metrics = self.calculate_ethical_metrics(open_facilities)
            
            return EthicalSolution(
                open_facilities=open_facilities,
                total_cost=evaluation['total_cost'],
                service_level=evaluation['service_level'],
                ethical_metrics=ethical_metrics,
                stakeholder_evaluations={},
                pareto_dominated=False
            )
        else:
            return None
    
    def _generate_environmental_solution(self) -> Optional[EthicalSolution]:
        """Generate environmentally-focused solution"""
        
        # Prioritize green facilities
        green_facilities = sorted(
            [f for f in self.facilities if f.environmental_footprint < 0.5],
            key=lambda f: f.environmental_footprint
        )
        
        open_facilities = set()
        evaluation = self.evaluate_basic_solution(open_facilities)
        
        # Add green facilities first
        for facility in green_facilities:
            open_facilities.add(facility.id)
            evaluation = self.evaluate_basic_solution(open_facilities)
            
        # Add other facilities if needed for service level
        if evaluation['service_level'] < 0.85:
            other_facilities = [f for f in self.facilities if f.id not in open_facilities]
            other_facilities.sort(key=lambda f: f.fixed_cost)
            
            for facility in other_facilities:
                open_facilities.add(facility.id)
                evaluation = self.evaluate_basic_solution(open_facilities)
                if evaluation['service_level'] >= 0.85:
                    break
        
        if open_facilities:
            ethical_metrics = self.calculate_ethical_metrics(open_facilities)
            
            return EthicalSolution(
                open_facilities=open_facilities,
                total_cost=evaluation['total_cost'],
                service_level=evaluation['service_level'],
                ethical_metrics=ethical_metrics,
                stakeholder_evaluations={},
                pareto_dominated=False
            )
        else:
            return None
    
    def _generate_balanced_solution(self) -> Optional[EthicalSolution]:
        """Generate balanced solution"""
        
        # Use weighted scoring for facility selection
        facility_scores = []
        for facility in self.facilities:
            # Calculate multi-criteria score
            cost_score = 1 - (facility.fixed_cost / max(f.fixed_cost for f in self.facilities))
            strategic_score = facility.community_impact
            community_score = facility.community_impact
            environmental_score = 1 - facility.environmental_footprint
            
            combined_score = (
                0.25 * cost_score +
                0.25 * strategic_score +
                0.25 * community_score +
                0.25 * environmental_score
            )
            
            facility_scores.append((facility.id, combined_score))
        
        # Sort by combined score (highest first)
        facility_scores.sort(key=lambda x: x[1], reverse=True)
        
        open_facilities = set()
        evaluation = self.evaluate_basic_solution(open_facilities)
        
        # Add facilities until service level is acceptable
        for facility_id, _ in facility_scores:
            if evaluation['service_level'] < 0.9:
                open_facilities.add(facility_id)
                evaluation = self.evaluate_basic_solution(open_facilities)
            else:
                break
        
        if open_facilities:
            ethical_metrics = self.calculate_ethical_metrics(open_facilities)
            
            return EthicalSolution(
                open_facilities=open_facilities,
                total_cost=evaluation['total_cost'],
                service_level=evaluation['service_level'],
                ethical_metrics=ethical_metrics,
                stakeholder_evaluations={},
                pareto_dominated=False
            )
        else:
            return None

print("Ethical CFLP Optimizer class defined successfully")
```
