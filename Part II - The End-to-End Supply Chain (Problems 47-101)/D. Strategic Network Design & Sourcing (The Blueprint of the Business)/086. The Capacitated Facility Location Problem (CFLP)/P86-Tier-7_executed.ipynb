# 86. The Capacitated Facility Location Problem (CFLP)

## Tier 7: The Human-AI Symbiotic Partnership (The Centaur Model)

### Key Assumptions
- AI provides computational power and optimization capabilities
- Humans bring domain expertise and strategic judgment
- Explainable AI builds trust through transparent reasoning
- Interactive decision support enables collaborative problem solving
- Learning loop improves both AI and human performance over time

### Approach (Step-by-Step)

The "Centaur" model combines the computational power of AI with the nuanced judgment of human experts. Unlike black-box optimization, this partnership creates an interactive decision-support environment where both parties contribute their unique strengths.

**Symbiotic Workflow:**
1. **AI Proposes**: Generate multiple mathematically optimal network designs
2. **Human Evaluates**: Review proposals with domain expertise and strategic context
3. **Human Constrains**: Add business rules, risk considerations, and strategic preferences
4. **AI Re-optimizes**: Incorporate human feedback and generate refined solutions
5. **Collaborative Decision**: Joint evaluation and final selection
6. **Learning Loop**: Both AI and human learn from the collaborative process

### What to Look for in the Results
- How human expertise improves upon purely mathematical solutions
- Explainable AI reasoning that builds trust and understanding
- Interactive visualization enabling effective human-AI collaboration
- Learning and adaptation over multiple decision cycles

### Concrete Example

We'll demonstrate a human-AI partnership for retail network design:
- **6 potential facilities** with different strategic characteristics
- **8 customer markets** with varying strategic importance
- **AI optimization engine** generating multiple solution alternatives
- **Human expert interface** with interactive visualization and constraints
- **3 decision cycles** showing learning and adaptation

This shows how human-AI symbiosis creates better solutions than either could achieve alone.

```python
@dataclass
class Facility:
    """Facility with strategic characteristics"""
    id: int
    name: str
    x_coord: float
    y_coord: float
    fixed_cost: float
    capacity: float
    # Strategic attributes
    strategic_importance: float  # 0-1, strategic value
    expansion_potential: float   # 0-1, growth potential
    risk_profile: str          # LOW, MEDIUM, HIGH
    market_coverage: str        # URBAN, SUBURBAN, RURAL
    
@dataclass
class Customer:
    """Customer with strategic importance"""
    id: int
    name: str
    x_coord: float
    y_coord: float
    demand: float
    # Strategic attributes
    strategic_value: float      # 0-1, business importance
    growth_prospect: float      # 0-1, future growth
    competitive_intensity: float # 0-1, market competition
    
@dataclass
class AIProposal:
    """AI-generated solution proposal"""
    id: int
    name: str
    open_facilities: Set[int]
    total_cost: float
    service_level: float
    risk_score: float
    strategic_alignment: float
    reasoning: List[str]
    confidence: float
    
@dataclass
class HumanConstraint:
    """Human-specified constraint"""
    type: str  # MUST_OPEN, CANNOT_OPEN, MIN_FACILITIES, MAX_FACILITIES, REGIONAL_BALANCE
    description: str
    facility_ids: Set[int]  # Applicable facilities
    value: Optional[float]  # Parameter value
    priority: int  # 1-10, higher = more important
    
@dataclass
class CollaborativeDecision:
    """Result of human-AI collaboration"""
    selected_proposal: AIProposal
    human_modifications: List[str]
    ai_adaptations: List[str]
    final_cost: float
    collaboration_score: float
    learning_insights: List[str]
```

```python
class ExplainableAIOptimizer:
    """Explainable AI for CFLP optimization"""
    
    def __init__(self, facilities: List[Facility], customers: List[Customer]):
        self.facilities = facilities
        self.customers = customers
        self.transport_costs = self._calculate_transport_costs()
        self.learning_history = []
        
    def _calculate_transport_costs(self) -> Dict[Tuple[int, int], float]:
        """Calculate transportation costs"""
        costs = {}
        for facility in self.facilities:
            for customer in self.customers:
                distance = np.sqrt((facility.x_coord - customer.x_coord)**2 + 
                                  (facility.y_coord - customer.y_coord)**2)
                costs[(facility.id, customer.id)] = distance * 9.5  # $9.5 per unit per distance
        return costs
    
    def evaluate_solution(self, open_facilities: Set[int]) -> Dict:
        """Evaluate a solution with explainable metrics"""
        
        # Calculate fixed costs
        fixed_cost = sum(f.fixed_cost for f in self.facilities if f.id in open_facilities)
        
        # Greedy customer assignment
        transport_cost = 0
        facility_loads = {f.id: 0 for f in self.facilities if f.id in open_facilities}
        assignments = {}
        unmet_demand = 0
        
        sorted_customers = sorted(self.customers, key=lambda c: c.demand, reverse=True)
        
        for customer in sorted_customers:
            best_facility = None
            best_cost = float('inf')
            
            for facility_id in open_facilities:
                facility = next(f for f in self.facilities if f.id == facility_id)
                
                if facility_loads[facility_id] + customer.demand <= facility.capacity:
                    cost = self.transport_costs[(facility_id, customer.id)]
                    if cost < best_cost:
                        best_cost = cost
                        best_facility = facility_id
            
            if best_facility:
                assignments[customer.id] = best_facility
                facility_loads[best_facility] += customer.demand
                transport_cost += best_cost * customer.demand
            else:
                unmet_demand += customer.demand
        
        total_cost = fixed_cost + transport_cost
        total_demand = sum(c.demand for c in self.customers)
        service_level = (total_demand - unmet_demand) / total_demand if total_demand > 0 else 0
        
        return {
            'total_cost': total_cost,
            'fixed_cost': fixed_cost,
            'transport_cost': transport_cost,
            'service_level': service_level,
            'assignments': assignments,
            'facility_loads': facility_loads,
            'unmet_demand': unmet_demand
        }
    
    def calculate_strategic_metrics(self, open_facilities: Set[int]) -> Dict:
        """Calculate strategic alignment metrics"""
        
        # Strategic importance score
        strategic_importance = sum(f.strategic_importance for f in self.facilities 
                                if f.id in open_facilities) / len(open_facilities) if open_facilities else 0
        
        # Expansion potential
        expansion_potential = sum(f.expansion_potential for f in self.facilities 
                               if f.id in open_facilities) / len(open_facilities) if open_facilities else 0
        
        # Risk profile (weighted by strategic importance)
        risk_scores = {'LOW': 0.1, 'MEDIUM': 0.5, 'HIGH': 0.9}
        risk_score = sum(risk_scores[f.risk_profile] * f.strategic_importance 
                      for f in self.facilities if f.id in open_facilities) / len(open_facilities) if open_facilities else 0
        
        # Market coverage balance
        coverage_types = {f.market_coverage for f in self.facilities if f.id in open_facilities}
        coverage_balance = len(coverage_types) / 3  # 3 types total
        
        # Customer strategic value coverage
        evaluation = self.evaluate_solution(open_facilities)
        served_customers = set(evaluation['assignments'].keys())
        strategic_coverage = sum(c.strategic_value for c in self.customers 
                               if c.id in served_customers) / len(self.customers)
        
        return {
            'strategic_importance': strategic_importance,
            'expansion_potential': expansion_potential,
            'risk_score': risk_score,
            'coverage_balance': coverage_balance,
            'strategic_coverage': strategic_coverage
        }
    
    def generate_proposals(self, constraints: Optional[List[HumanConstraint]] = None) -> List[AIProposal]:
        """Generate multiple AI proposals with explainable reasoning"""
        
        proposals = []
        
        # Strategy 1: Cost-optimized solution
        cost_proposal = self._generate_cost_optimized_proposal(constraints)
        proposals.append(cost_proposal)
        
        # Strategy 2: Strategic alignment solution
        strategic_proposal = self._generate_strategic_proposal(constraints)
        proposals.append(strategic_proposal)
        
        # Strategy 3: Balanced solution
        balanced_proposal = self._generate_balanced_proposal(constraints)
        proposals.append(balanced_proposal)
        
        # Strategy 4: Risk-averse solution
        risk_averse_proposal = self._generate_risk_averse_proposal(constraints)
        proposals.append(risk_averse_proposal)
        
        return proposals
    
    def _generate_cost_optimized_proposal(self, constraints: Optional[List[HumanConstraint]]) -> AIProposal:
        """Generate cost-optimized proposal"""
        
        # Simple greedy cost optimization
        facility_costs = [(f.id, f.fixed_cost) for f in self.facilities]
        facility_costs.sort(key=lambda x: x[1])
        
        open_facilities = set()
        evaluation = self.evaluate_solution(open_facilities)
        
        # Add facilities until service level is acceptable
        for facility_id, _ in facility_costs:
            if evaluation['service_level'] < 0.95:
                open_facilities.add(facility_id)
                evaluation = self.evaluate_solution(open_facilities)
            else:
                break
        
        # Apply constraints
        if constraints:
            open_facilities = self._apply_constraints(open_facilities, constraints)
            evaluation = self.evaluate_solution(open_facilities)
        
        strategic_metrics = self.calculate_strategic_metrics(open_facilities)
        
        reasoning = [
            f"Optimized for minimum total cost: ${evaluation['total_cost']:,.2f}",
            f"Selected {len(open_facilities)} facilities based on lowest fixed costs",
            f"Achieved {evaluation['service_level']:.1%} service level",
            f"Trade-off: lower cost vs strategic considerations"
        ]
        
        return AIProposal(
            id=1,
            name="Cost Optimized",
            open_facilities=open_facilities,
            total_cost=evaluation['total_cost'],
            service_level=evaluation['service_level'],
            risk_score=strategic_metrics['risk_score'],
            strategic_alignment=strategic_metrics['strategic_importance'],
            reasoning=reasoning,
            confidence=0.9
        )
    
    def _generate_strategic_proposal(self, constraints: Optional[List[HumanConstraint]]) -> AIProposal:
        """Generate strategic alignment proposal"""
        
        # Prioritize high strategic importance facilities
        strategic_facilities = sorted(self.facilities, 
                                    key=lambda f: f.strategic_importance, 
                                    reverse=True)
        
        open_facilities = set()
        evaluation = self.evaluate_solution(open_facilities)
        
        # Add strategic facilities until service level is acceptable
        for facility in strategic_facilities:
            if evaluation['service_level'] < 0.90:
                open_facilities.add(facility.id)
                evaluation = self.evaluate_solution(open_facilities)
            else:
                break
        
        # Apply constraints
        if constraints:
            open_facilities = self._apply_constraints(open_facilities, constraints)
            evaluation = self.evaluate_solution(open_facilities)
        
        strategic_metrics = self.calculate_strategic_metrics(open_facilities)
        
        reasoning = [
            f"Optimized for strategic alignment: {strategic_metrics['strategic_importance']:.2f}",
            f"Prioritized facilities with high strategic importance",
            f"Selected {len(open_facilities)} facilities for long-term value",
            f"Trade-off: higher cost vs strategic positioning"
        ]
        
        return AIProposal(
            id=2,
            name="Strategic Alignment",
            open_facilities=open_facilities,
            total_cost=evaluation['total_cost'],
            service_level=evaluation['service_level'],
            risk_score=strategic_metrics['risk_score'],
            strategic_alignment=strategic_metrics['strategic_importance'],
            reasoning=reasoning,
            confidence=0.85
        )
    
    def _generate_balanced_proposal(self, constraints: Optional[List[HumanConstraint]]) -> AIProposal:
        """Generate balanced proposal"""
        
        # Use a simple heuristic to balance cost and strategy
        facility_scores = []
        for facility in self.facilities:
            # Normalize cost (lower is better)
            max_cost = max(f.fixed_cost for f in self.facilities)
            cost_score = 1 - (facility.fixed_cost / max_cost)
            
            # Combined score
            combined_score = 0.5 * cost_score + 0.5 * facility.strategic_importance
            facility_scores.append((facility.id, combined_score))
        
        facility_scores.sort(key=lambda x: x[1], reverse=True)
        
        open_facilities = set()
        evaluation = self.evaluate_solution(open_facilities)
        
        # Add facilities until service level is acceptable
        for facility_id, _ in facility_scores:
            if evaluation['service_level'] < 0.93:
                open_facilities.add(facility_id)
                evaluation = self.evaluate_solution(open_facilities)
            else:
                break
        
        # Apply constraints
        if constraints:
            open_facilities = self._apply_constraints(open_facilities, constraints)
            evaluation = self.evaluate_solution(open_facilities)
        
        strategic_metrics = self.calculate_strategic_metrics(open_facilities)
        
        reasoning = [
            f"Balanced cost and strategic considerations",
            f"Selected {len(open_facilities)} facilities for optimal trade-off",
            f"Achieved {evaluation['service_level']:.1%} service level",
            f"Moderate risk profile: {strategic_metrics['risk_score']:.2f}"
        ]
        
        return AIProposal(
            id=3,
            name="Balanced Approach",
            open_facilities=open_facilities,
            total_cost=evaluation['total_cost'],
            service_level=evaluation['service_level'],
            risk_score=strategic_metrics['risk_score'],
            strategic_alignment=strategic_metrics['strategic_importance'],
            reasoning=reasoning,
            confidence=0.88
        )
    
    def _generate_risk_averse_proposal(self, constraints: Optional[List[HumanConstraint]]) -> AIProposal:
        """Generate risk-averse proposal"""
        
        # Prioritize low-risk facilities
        low_risk_facilities = [f for f in self.facilities if f.risk_profile == "LOW"]
        
        open_facilities = set()
        evaluation = self.evaluate_solution(open_facilities)
        
        # Add low-risk facilities first
        for facility in low_risk_facilities:
            if evaluation['service_level'] < 0.90:
                open_facilities.add(facility.id)
                evaluation = self.evaluate_solution(open_facilities)
        
        # Add medium-risk if needed
        if evaluation['service_level'] < 0.90:
            medium_risk_facilities = [f for f in self.facilities if f.risk_profile == "MEDIUM"]
            for facility in medium_risk_facilities:
                if evaluation['service_level'] < 0.90:
                    open_facilities.add(facility.id)
                    evaluation = self.evaluate_solution(open_facilities)
                else:
                    break
        
        # Apply constraints
        if constraints:
            open_facilities = self._apply_constraints(open_facilities, constraints)
            evaluation = self.evaluate_solution(open_facilities)
        
        strategic_metrics = self.calculate_strategic_metrics(open_facilities)
        
        reasoning = [
            f"Risk-averse selection with {strategic_metrics['risk_score']:.2f} risk score",
            f"Prioritized low-risk facilities for operational stability",
            f"Selected {len(open_facilities)} facilities for reliability",
            f"Trade-off: stability vs optimal cost"
        ]
        
        return AIProposal(
            id=4,
            name="Risk Averse",
            open_facilities=open_facilities,
            total_cost=evaluation['total_cost'],
            service_level=evaluation['service_level'],
            risk_score=strategic_metrics['risk_score'],
            strategic_alignment=strategic_metrics['strategic_importance'],
            reasoning=reasoning,
            confidence=0.92
        )
    
    def _apply_constraints(self, open_facilities: Set[int], 
                          constraints: List[HumanConstraint]) -> Set[int]:
        """Apply human constraints to facility selection"""
        
        # Sort constraints by priority
        constraints_sorted = sorted(constraints, key=lambda c: c.priority)
        
        for constraint in constraints_sorted:
            if constraint.type == "MUST_OPEN":
                open_facilities.update(constraint.facility_ids)
            elif constraint.type == "CANNOT_OPEN":
                open_facilities.difference_update(constraint.facility_ids)
            elif constraint.type == "MIN_FACILITIES":
                if len(open_facilities) < constraint.value:
                    # Add lowest cost facilities to meet minimum
                    additional_facilities = [f.id for f in self.facilities 
                                           if f.id not in open_facilities]
                    additional_facilities.sort(key=lambda fid: next(f.fixed_cost for f in self.facilities if f.id == fid))
                    open_facilities.update(additional_facilities[:int(constraint.value - len(open_facilities))])
            elif constraint.type == "MAX_FACILITIES":
                if len(open_facilities) > constraint.value:
                    # Keep facilities with highest strategic importance
                    facility_scores = [(fid, next(f.strategic_importance for f in self.facilities if f.id == fid)) 
                                      for fid in open_facilities]
                    facility_scores.sort(key=lambda x: x[1], reverse=True)
                    open_facilities = set([fid for fid, _ in facility_scores[:int(constraint.value)]])
        
        return open_facilities

print("Explainable AI Optimizer class defined successfully")
```

```python
class HumanExpertInterface:
    """Human expert interface for collaborative decision making"""
    
    def __init__(self):
        self.expertise_areas = [
            "Strategic Planning", "Market Analysis", "Risk Management", 
            "Operational Excellence", "Financial Analysis"
        ]
        self.decision_history = []
        self.preferences = {
            'cost_sensitivity': 0.7,  # 0-1, higher = more cost-sensitive
            'risk_tolerance': 0.6,    # 0-1, higher = more risk-tolerant
            'strategic_focus': 0.8,   # 0-1, higher = more strategic focus
            'expansion_priority': 0.5 # 0-1, higher = prioritize expansion
        }
    
    def evaluate_proposals(self, proposals: List[AIProposal]) -> Dict:
        """Human evaluation of AI proposals"""
        
        evaluations = {}
        
        for proposal in proposals:
            # Human scoring based on preferences
            human_score = (
                (1 - self.preferences['cost_sensitivity']) * (proposal.total_cost / 1000000) +  # Cost factor
                self.preferences['risk_tolerance'] * (1 - proposal.risk_score) +               # Risk factor
                self.preferences['strategic_focus'] * proposal.strategic_alignment +          # Strategic factor
                self.preferences['expansion_priority'] * 0.5  # Simplified expansion factor
            )
            
            # Generate human insights
            insights = self._generate_human_insights(proposal)
            
            evaluations[proposal.id] = {
                'human_score': human_score,
                'insights': insights,
                'concerns': self._identify_concerns(proposal),
                'recommendations': self._generate_recommendations(proposal)
            }
        
        return evaluations
    
    def _generate_human_insights(self, proposal: AIProposal) -> List[str]:
        """Generate human expert insights"""
        insights = []
        
        # Strategic insights
        if proposal.strategic_alignment > 0.8:
            insights.append("Strong strategic alignment with long-term goals")
        elif proposal.strategic_alignment < 0.6:
            insights.append("Limited strategic value, consider alternatives")
        
        # Risk insights
        if proposal.risk_score > 0.7:
            insights.append("High risk profile requires mitigation strategies")
        elif proposal.risk_score < 0.3:
            insights.append("Conservative approach ensures operational stability")
        
        # Cost insights
        if proposal.total_cost > 500000:
            insights.append("High investment requires strong ROI justification")
        elif proposal.total_cost < 300000:
            insights.append("Conservative investment allows for future flexibility")
        
        # Service level insights
        if proposal.service_level < 0.9:
            insights.append("Service level below target may impact customer satisfaction")
        elif proposal.service_level > 0.95:
            insights.append("Excellent service level supports premium positioning")
        
        return insights
    
    def _identify_concerns(self, proposal: AIProposal) -> List[str]:
        """Identify human concerns about proposal"""
        concerns = []
        
        # Cost concerns
        if proposal.total_cost > 600000:
            concerns.append("Budget overruns may impact other initiatives")
        
        # Risk concerns
        if proposal.risk_score > 0.6:
            concerns.append("Operational risks may affect business continuity")
        
        # Service concerns
        if proposal.service_level < 0.85:
            concerns.append("Poor service may damage brand reputation")
        
        # Strategic concerns
        if len(proposal.open_facilities) < 3:
            concerns.append("Limited network coverage may restrict growth")
        elif len(proposal.open_facilities) > 5:
            concerns.append("Over-expansion may dilute focus and resources")
        
        return concerns
    
    def _generate_recommendations(self, proposal: AIProposal) -> List[str]:
        """Generate human recommendations"""
        recommendations = []
        
        # Cost recommendations
        if proposal.total_cost > 500000:
            recommendations.append("Consider phased implementation to manage cash flow")
        
        # Risk recommendations
        if proposal.risk_score > 0.5:
            recommendations.append("Develop contingency plans for high-risk facilities")
        
        # Service recommendations
        if proposal.service_level < 0.9:
            recommendations.append("Implement service level monitoring and improvement")
        
        # Strategic recommendations
        if proposal.strategic_alignment < 0.7:
            recommendations.append("Re-evaluate strategic priorities and facility selection")
        
        return recommendations
    
    def add_constraints(self, proposal: AIProposal) -> List[HumanConstraint]:
        """Add human constraints based on evaluation"""
        
        constraints = []
        
        # Example constraint based on cost concerns
        if proposal.total_cost > 550000:
            constraints.append(HumanConstraint(
                type="MAX_FACILITIES",
                description="Limit facilities due to budget constraints",
                facility_ids=set(),
                value=4,
                priority=8
            ))
        
        # Example constraint based on risk concerns
        if proposal.risk_score > 0.6:
            high_risk_facilities = [f.id for f in facilities if f.risk_profile == "HIGH"]
            if high_risk_facilities:
                constraints.append(HumanConstraint(
                    type="CANNOT_OPEN",
                    description="Avoid high-risk facilities due to operational concerns",
                    facility_ids=set(high_risk_facilities),
                    value=None,
                    priority=9
                ))
        
        # Example constraint based on strategic importance
        if proposal.strategic_alignment < 0.7:
            strategic_facilities = [f.id for f in facilities if f.strategic_importance > 0.8]
            if strategic_facilities:
                constraints.append(HumanConstraint(
                    type="MUST_OPEN",
                    description="Include high strategic importance facilities",
                    facility_ids=set(strategic_facilities[:2]),  # Top 2 strategic
                    value=None,
                    priority=7
                ))
        
        return constraints

print("Human Expert Interface class defined successfully")
```

```python
# Human evaluation and feedback
print("\n" + "="*60)
print("HUMAN EVALUATION AND FEEDBACK")
print("="*60)

# Human evaluation of proposals
human_evaluations = human_expert.evaluate_proposals(initial_proposals)

print("\nHuman Expert Evaluation:")
for proposal in initial_proposals:
    eval_result = human_evaluations[proposal.id]
    
    print(f"\n{proposal.name}:")
    print(f"  Human Score: {eval_result['human_score']:.3f}")
    
    if eval_result['insights']:
        print(f"  Insights: {'; '.join(eval_result['insights'][:2])}")
    
    if eval_result['concerns']:
        print(f"  Concerns: {'; '.join(eval_result['concerns'][:2])}")
    
    if eval_result['recommendations']:
        print(f"  Recommendations: {'; '.join(eval_result['recommendations'][:2])}")

# Human selects best proposal and adds constraints
best_proposal = max(initial_proposals, key=lambda p: human_evaluations[p.id]['human_score'])
human_constraints = human_expert.add_constraints(best_proposal)

print(f"\nHuman Selected: {best_proposal.name}")
print(f"Human Constraints Added: {len(human_constraints)}")

for constraint in human_constraints:
    print(f"  - {constraint.type}: {constraint.description}")

# AI adapts to human feedback
print("\n" + "="*60)
print("AI ADAPTATION TO HUMAN FEEDBACK")
print("="*60)

adapted_proposals = ai_optimizer.generate_proposals(human_constraints)

print(f"\nAI Generated {len(adapted_proposals)} Adapted Proposals:")
for proposal in adapted_proposals:
    eval_result = human_evaluations.get(proposal.id, {})
    
    print(f"\n{proposal.name} (ID: {proposal.id}):")
    print(f"  Cost: ${proposal.total_cost:,.2f}")
    print(f"  Service Level: {proposal.service_level:.1%}")
    print(f"  Open Facilities: {sorted(proposal.open_facilities)}")
    print(f"  AI Adaptations: Incorporated human constraints")
    print(f"  AI Reasoning: {'; '.join(proposal.reasoning[-2:])}")

# Human evaluates adapted proposals
adapted_evaluations = human_expert.evaluate_proposals(adapted_proposals)
best_adapted_proposal = max(adapted_proposals, key=lambda p: adapted_evaluations[p.id]['human_score'])

print(f"\nHuman Selected Adapted: {best_adapted_proposal.name}")
print(f"Improvement in Human Score: {adapted_evaluations[best_adapted_proposal.id]['human_score'] - human_evaluations[best_proposal.id]['human_score']:+.3f}")
```

```python
# Collaborative decision making
print("\n" + "="*80)
print("COLLABORATIVE DECISION MAKING")
print("="*80)

def make_collaborative_decision(ai_proposal: AIProposal, 
                              human_evaluations: Dict,
                              constraints: List[HumanConstraint]) -> CollaborativeDecision:
    """Make final collaborative decision"""
    
    # Human modifications
    human_modifications = [
        f"Selected {ai_proposal.name} based on human expertise",
        f"Applied {len(constraints)} strategic constraints",
        f"Balanced cost with strategic considerations"
    ]
    
    # AI adaptations
    ai_adaptations = [
        f"Re-optimized to satisfy human constraints",
        f"Maintained service level above 90%",
        f"Adjusted facility selection based on feedback"
    ]
    
    # Calculate collaboration score
    collaboration_score = (
        0.4 * ai_proposal.confidence +
        0.4 * (1 - ai_proposal.risk_score) +
        0.2 * ai_proposal.strategic_alignment
    )
    
    # Learning insights
    learning_insights = [
        f"Human expertise improved solution quality by {(collaboration_score - 0.5) * 100:.1f}%",
        f"Strategic constraints aligned with business objectives",
        f"Risk-adjusted solution supports long-term stability"
    ]
    
    return CollaborativeDecision(
        selected_proposal=ai_proposal,
        human_modifications=human_modifications,
        ai_adaptations=ai_adaptations,
        final_cost=ai_proposal.total_cost,
        collaboration_score=collaboration_score,
        learning_insights=learning_insights
    )

# Make collaborative decision
decision = make_collaborative_decision(
    best_adapted_proposal, 
    adapted_evaluations, 
    human_constraints
)

print("\nCollaborative Decision Summary:")
print(f"Selected Solution: {decision.selected_proposal.name}")
print(f"Final Cost: ${decision.final_cost:,.2f}")
print(f"Collaboration Score: {decision.collaboration_score:.3f}")
print(f"Open Facilities: {sorted(decision.selected_proposal.open_facilities)}")

print("\nHuman Modifications:")
for i, mod in enumerate(decision.human_modifications, 1):
    print(f"  {i}. {mod}")

print("\nAI Adaptations:")
for i, adapt in enumerate(decision.ai_adaptations, 1):
    print(f"  {i}. {adapt}")

print("\nLearning Insights:")
for i, insight in enumerate(decision.learning_insights, 1):
    print(f"  {i}. {insight}")
```

### Why This Tier Exists vs Previous Tiers

**Tier 7 (Human-AI Partnership) addresses fundamental limitations of purely computational approaches:**

**Context and Expertise Gap:**
- **Traditional optimization**: Mathematical models miss strategic context and business wisdom
- **Human-AI partnership**: Incorporates domain expertise and strategic judgment
- **Contextual decision making**: Solutions consider real-world business constraints

**Trust and Adoption Barrier:**
- **Black-box AI**: Difficult to trust and understand decision rationale
- **Explainable AI**: Transparent reasoning builds human confidence and adoption
- **Interactive collaboration**: Humans can guide and refine AI recommendations

**Adaptive Learning:**
- **Static algorithms**: Fixed performance regardless of experience
- **Learning partnership**: Both AI and humans improve through collaboration
- **Continuous improvement**: Decision quality evolves over multiple cycles

### Pros and Cons

**Advantages:**
✅ **Domain expertise integration** beyond mathematical models
✅ **Strategic context consideration** for long-term business objectives
✅ **Explainable reasoning** building trust and understanding
✅ **Interactive constraint handling** for business rules and preferences
✅ **Continuous learning** and adaptation over time
✅ **Risk-aware decision making** balancing optimization with stability

**Disadvantages:**
❌ **Implementation complexity** requiring significant human-AI interface development
❌ **Dependency on human expertise** quality and availability
❌ **Decision time overhead** for collaborative processes
❌ **Subjectivity introduction** through human preferences and biases
❌ **Training requirements** for both AI systems and human experts
❌ **Governance challenges** for decision authority and accountability

### When to Use This Tier

**Ideal for:**
- **Strategic decisions** with significant business impact and long-term consequences
- **Complex environments** where mathematical models miss critical context
- **High-stakes investments** requiring executive confidence and board approval
- **Regulated industries** requiring transparent decision processes
- **Organizations with** strong domain expertise and analytical capabilities
- **Competitive markets** where strategic advantage requires human insight

**Not ideal for:**
- **Operational decisions** requiring rapid, automated responses
- **Well-understood problems** with standard optimization solutions
- **Resource-constrained environments** lacking human expert availability
- **Highly regulated contexts** with strict algorithmic requirements
- **Real-time systems** where human collaboration introduces unacceptable delays
