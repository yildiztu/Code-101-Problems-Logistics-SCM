```python
from itertools import product
from itertools import combinations

# Import required libraries for digital twin simulation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import time
from datetime import datetime, timedelta
from collections import defaultdict
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
random.seed(42)

# Set style for professional visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
try:
    # Define digital twin architecture components
    class SupplierDigitalTwin:
        """Digital twin for supplier operations"""
        def __init__(self, name, location, capacity, cost, risk, quality):
            self.name = name
            self.location = location
            self.base_capacity = capacity
            self.current_capacity = capacity
            self.base_cost = cost
            self.current_cost = cost
            self.risk_factor = risk
            self.quality_score = quality
            
            # Operational status
            self.operational_status = 1.0  # 1.0 = fully operational
            self.production_efficiency = 1.0
            self.quality_control_score = 1.0
            
            # IoT sensor data
            self.sensor_data = {
                'production_rate': [],
                'quality_metrics': [],
                'equipment_status': [],
                'inventory_levels': []
                'delivery_performance': []
                'operational_uptime': []
            }
            
            # Predictive analytics
            self.disruption_probability = 0.0
            self.capacity_forecast = []
            self.cost_trend = []
        
        def update_sensor_data(self, timestamp):
            """Simulate IoT sensor data updates"""
            # Simulate sensor readings with some noise
            production_rate = self.current_capacity * self.operational_status * np.random.normal(1.0, 0.05)
            quality_metric = self.quality_score * self.quality_control_score * np.random.normal(1.0, 0.02)
            equipment_status = self.operational_status * np.random.normal(1.0, 0.03)
            inventory_level = np.random.uniform(0.3, 0.8) * self.current_capacity
            delivery_performance = np.random.uniform(0.85, 0.98) * self.operational_status
            operational_uptime = np.random.uniform(0.92, 0.99)
            
            # Store sensor data
            self.sensor_data['production_rate'].append((timestamp, production_rate))
            self.sensor_data['quality_metrics'].append((timestamp, quality_metric))
            self.sensor_data['equipment_status'].append((timestamp, equipment_status))
            self.sensor_data['inventory_levels'].append((timestamp, inventory_level))
            self.sensor_data['delivery_performance'].append((timestamp, delivery_performance))
            self.sensor_data['operational_uptime'].append((timestamp, operational_uptime))
        
        def apply_disruption(self, disruption_type, severity):
            """Apply disruption to supplier operations"""
            if disruption_type == 'tariff_increase':
                self.current_cost = self.base_cost * (1 + severity)
            elif disruption_type == 'capacity_reduction':
                self.current_capacity = self.base_capacity * (1 - severity)
            elif disruption_type == 'quality_issue':
                self.quality_control_score = 1.0 - severity
            elif disruption_type == 'operational_issue':
                self.operational_status = 1.0 - severity
            
            self.disruption_probability = severity
    
    class LogisticsDigitalTwin:
        """Digital twin for logistics network"""
        def __init__(self):
            self.routes = {}
            self.ports = {}
            self.transportation_costs = {}
            self.lead_times = {}
            
            # Real-time tracking
            self.shipment_tracking = {}
            self.port_congestion = {}
            self.weather_impacts = {}
            
            # Predictive analytics
            self.congestion_forecasts = {}
            self.delay_predictions = {}
        
        def update_network_status(self, timestamp):
            """Update logistics network status"""
            # Simulate port congestion
            for port in self.ports:
                congestion = np.random.beta(2, 5)  # Beta distribution for congestion
                self.port_congestion[port] = congestion
                
                # Update congestion forecasts
                if port not in self.congestion_forecasts:
                    self.congestion_forecasts[port] = []
                self.congestion_forecasts[port].append((timestamp, congestion))
    
    class MarketDemandTwin:
        """Digital twin for market demand and conditions"""
        def __init__(self):
            self.demand_forecasts = {}
            self.market_conditions = {}
            self.competitor_actions = {}
            self.customer_preferences = {}
            
            # Real-time data
            self.current_demand = {}
            self.price_elasticity = {}
            self.market_trends = {}
        
        def update_market_data(self, timestamp):
            """Update market demand and conditions"""
            # Simulate market dynamics
            for product in self.current_demand:
                # Add some randomness to demand
                demand_change = np.random.normal(0, 0.05)  # 5% standard deviation
                self.current_demand[product] *= (1 + demand_change)
                
                # Update demand forecasts
                if product not in self.demand_forecasts:
                    self.demand_forecasts[product] = []
                self.demand_forecasts[product].append((timestamp, self.current_demand[product]))
    
    print("Digital twin components initialized...")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
class GlobalSourcingDigitalTwin:
    """Integrated digital twin for global sourcing ecosystem"""
    def __init__(self):
        # Initialize digital twin components
        self.suppliers = {}
        self.logistics = LogisticsDigitalTwin()
        self.market = MarketDemandTwin()
        
        # System state
        self.current_time = datetime(2024, 1, 1)
        self.simulation_speed = 1000  # 1000x real-time
        
        # Optimization engine
        self.current_allocation = {}
        self.optimization_results = {}
        
        # KPI monitoring
        self.kpi_history = {
            'total_cost': [],
            'supply_risk': [],
            'quality_score': [],
            'delivery_performance': [],
            'inventory_levels': [],
            'disruption_events': [],
        }
        
        # Scenario analysis
        self.scenario_engine = ScenarioEngine()
        self.current_scenarios = []
        
        # Predictive analytics
        self.risk_predictions = {}
        self.demand_predictions = {}
        self.cost_projections = {}
    
    def initialize_suppliers(self):
        """Initialize supplier digital twins"""
        # China suppliers
        self.suppliers['China_Electronics'] = SupplierDigitalTwin(
            'China_Electronics', 'China', 50000, 25, 0.8, 0.7
        )
        self.suppliers['China_Manufacturing'] = SupplierDigitalTwin(
            'China_Manufacturing', 'China', 45000, 35, 0.8, 0.75
        )
        
        # Mexico suppliers
        self.suppliers['Mexico_Auto'] = SupplierDigitalTwin(
            'Mexico_Auto', 'Mexico', 30000, 35, 0.4, 0.8
        )
        self.suppliers['Mexico_Electronics'] = SupplierDigitalTwin(
            'Mexico_Electronics', 'Mexico', 25000, 40, 0.4, 0.82
        )
        
        # US suppliers
        self.suppliers['US_Premium'] = SupplierDigitalTwin(
            'US_Premium', 'US', 25000, 50, 0.1, 0.95
        )
        self.suppliers['US_Tech'] = SupplierDigitalTwin(
            'US_Tech', 'US', 20000, 55, 0.1, 0.97
        )
        
        # Initialize logistics network
        self.logistics.ports = {
            'Shanghai': {'capacity': 1000000, 'current_utilization': 0.7},
            'Los_Angeles': {'capacity': 800000, 'current_utilization': 0.6},
            'New_York': {'capacity': 600000, 'current_utilization': 0.5},
            'Rotterdam': {'capacity': 900000, 'current_utilization': 0.65},
        }
        
        # Initialize market demand
        self.market.current_demand = {
            'Microprocessor': 10000,
            'Display': 8000,
            'Battery': 12000,
            'Sensor': 5000,
        }
    
    def update_system_state(self, time_step_hours=1):
        """Update all digital twin components"""
        # Update current time
        self.current_time += timedelta(hours=time_step_hours)
        
        # Update suppliers
        for supplier in self.suppliers.values():
            supplier.update_sensor_data(self.current_time)
        
        # Update logistics
        self.logistics.update_network_status(self.current_time)
        
        # Update market
        self.market.update_market_data(self.current_time)
        
        # Update predictive analytics
        self.update_predictions()
    
    def update_predictions(self):
        """Update predictive analytics"""
        # Predict disruptions based on sensor data
        for supplier_name, supplier in self.suppliers.items():
            # Analyze recent sensor data for disruption indicators
            recent_uptime = [data[1] for data in supplier.sensor_data['operational_uptime'][-10:]]
            if recent_uptime:
                avg_uptime = np.mean(recent_uptime)
                if avg_uptime < 0.95:
                    supplier.disruption_probability = 1.0 - avg_uptime
                else:
                    supplier.disruption_probability = 0.01
    
    def optimize_sourcing_allocation(self):
        """Optimize sourcing allocation based on current system state"""
        # Get current market demand
        demands = self.market.current_demand
        
        # Get supplier capabilities and costs
        supplier_capabilities = {}
        supplier_costs = {}
        supplier_risks = {}
        supplier_qualities = {}
        
        for supplier_name, supplier in self.suppliers.items():
            supplier_capabilities[supplier_name] = supplier.current_capacity * supplier.operational_status
            supplier_costs[supplier_name] = supplier.current_cost
            supplier_risks[supplier_name] = supplier.risk_factor * (2 - supplier.operational_status)
            supplier_qualities[supplier_name] = supplier.quality_score * supplier.quality_control_score
        
        # Simple optimization (would use more sophisticated algorithm in practice)
        allocation = {}
        
        for product in demands:
            allocation[product] = {}
            demand = demands[product]
            
            # Calculate weighted scores for each supplier
            scores = {}
            for supplier_name in supplier_capabilities:
                # Lower cost, lower risk, higher quality = better score
                cost_score = 1.0 / supplier_costs[supplier_name]
                risk_score = 1.0 / (1.0 + supplier_risks[supplier_name])
                quality_score = supplier_qualities[supplier_name]
                
                # Weighted combination
                scores[supplier_name] = 0.6 * cost_score + 0.3 * risk_score + 0.1 * quality_score
            
            # Allocate based on scores
            total_score = sum(scores.values())
            for supplier_name in scores:
                allocation[product][supplier_name] = demand * (scores[supplier_name] / total_score)
        
        self.current_allocation = allocation
        return allocation
    
    def calculate_kpis(self):
        """Calculate current KPIs"""
        if not self.current_allocation:
            return
        
        # Calculate total cost
        total_cost = 0
        total_risk = 0
        total_quality = 0
        
        for product, allocation in self.current_allocation.items():
            for supplier_name, quantity in allocation.items():
                if supplier_name in self.suppliers:
                    supplier = self.suppliers[supplier_name]
                    
                    # Cost components
                    total_cost += supplier.current_cost * quantity
                    total_cost += supplier.base_cost * 400  # Fixed cost
                    
                    # Risk and quality
                    total_risk += supplier.risk_factor * quantity
                    total_quality += supplier.quality_score * quantity
        
        # Calculate delivery performance
        delivery_performance = np.mean([s.operational_status for s in self.suppliers.values()])
        
        # Calculate inventory levels
        total_inventory = sum([np.mean([data[1] for data in s.sensor_data['inventory_levels'][-5:]]) 
                              for s in self.suppliers.values() if s.sensor_data['inventory_levels']])
        
        # Store KPIs
        self.kpi_history['total_cost'].append((self.current_time, total_cost))
        self.kpi_history['supply_risk'].append((self.current_time, total_risk))
        self.kpi_history['quality_score'].append((self.current_time, total_quality))
        self.kpi_history['delivery_performance'].append((self.current_time, delivery_performance))
        self.kpi_history['inventory_levels'].append((self.current_time, total_inventory))
        
        return {
            'total_cost': total_cost,
            'supply_risk': total_risk,
            'quality_score': total_quality,
            'delivery_performance': delivery_performance,
            'inventory_levels': total_inventory
        }

print("Global sourcing digital twin initialized...")
```

```python
class ScenarioEngine:
    """Scenario analysis engine for what-if simulations"""
    def __init__(self):
        self.scenario_library = {
            'tariff_increase': self.simulate_tariff_increase,
            'port_closure': self.simulate_port_closure,
            'currency_devaluation': self.simulate_currency_devaluation,
            'supply_disruption': self.simulate_supply_disruption,
            'demand_surge': self.simulate_demand_surge,
            'quality_crisis': self.simulate_quality_crisis,
        }
        
        self.scenario_results = {}
    
    def simulate_tariff_increase(self, digital_twin, affected_suppliers, tariff_rate):
        """Simulate tariff increase scenario"""
        print(f"\n=== TARIFF INCREASE SCENARIO ===")
        print(f"Affected suppliers: {affected_suppliers}")
        print(f"Tariff rate: {tariff_rate * 100:.0f}%")
        
        # Store original state
        original_costs = {}
        for supplier_name in affected_suppliers:
            if supplier_name in digital_twin.suppliers:
                original_costs[supplier_name] = digital_twin.suppliers[supplier_name].current_cost
        
        # Apply tariff
        for supplier_name in affected_suppliers:
            if supplier_name in digital_twin.suppliers:
                digital_twin.suppliers[supplier_name].apply_disruption('tariff_increase', tariff_rate)
        
        # Re-optimize allocation
        new_allocation = digital_twin.optimize_sourcing_allocation()
        new_kpis = digital_twin.calculate_kpis()
        
        # Calculate impact
        cost_increase = new_kpis['total_cost'] - sum(original_costs.values()) * 10000  # Rough estimate
        
        # Analyze allocation changes
        allocation_shift = self.analyze_allocation_shift(digital_twin.current_allocation, affected_suppliers)
        
        # Restore original state
        for supplier_name, original_cost in original_costs.items():
            digital_twin.suppliers[supplier_name].current_cost = original_cost
        
        return {
            'scenario_type': 'tariff_increase',
            'cost_impact': cost_increase,
            'allocation_shift': allocation_shift,
            'new_allocation': new_allocation,
            'new_kpis': new_kpis
        }
    
    def simulate_port_closure(self, digital_twin, closed_port, duration_days):
        """Simulate port closure scenario"""
        print(f"\n=== PORT CLOSURE SCENARIO ===")
        print(f"Closed port: {closed_port}")
        print(f"Duration: {duration_days} days")
        
        # Increase logistics costs and lead times for affected routes
        logistics_impact = duration_days * 0.1  # 10% cost increase per day
        
        # Simulate rerouting
        rerouting_cost = 50000 * duration_days  # $50K per day
        delivery_delay = duration_days * 0.5  # 50% of shipments delayed
        
        return {
            'scenario_type': 'port_closure',
            'logistics_impact': logistics_impact,
            'rerouting_cost': rerouting_cost,
            'delivery_delay': delivery_delay,
            'customer_impact': max(0, 100 - delivery_delay * 10)  # Customer satisfaction
        }
    
    def simulate_currency_devaluation(self, digital_twin, currency, devaluation_rate):
        """Simulate currency devaluation scenario"""
        print(f"\n=== CURRENCY DEVALUATION SCENARIO ===")
        print(f"Currency: {currency}")
        print(f"Devaluation: {devaluation_rate * 100:.0f}%")
        
        # Calculate financial impact
        affected_suppliers = [s for s in digital_twin.suppliers.keys() if currency in s.lower()]
        
        total_exposure = 0
        for supplier_name in affected_suppliers:
            if supplier_name in digital_twin.suppliers:
                supplier = digital_twin.suppliers[supplier_name]
                # Estimate monthly exposure
                monthly_exposure = supplier.current_capacity * supplier.current_cost
                total_exposure += monthly_exposure
        
        # Calculate hedging benefit
        hedging_benefit = total_exposure * devaluation_rate * 0.8  # 80% effective hedging
        
        return {
            'scenario_type': 'currency_devaluation',
            'total_exposure': total_exposure,
            'hedging_benefit': hedging_benefit,
            'net_impact': total_exposure * devaluation_rate - hedging_benefit,
            'affected_suppliers': affected_suppliers
        }
    
    def analyze_allocation_shift(self, new_allocation, affected_suppliers):
        """Analyze how allocation shifted away from affected suppliers"""
        total_shift = 0
        
        for product, allocation in new_allocation.items():
            for supplier_name, quantity in allocation.items():
                if supplier_name in affected_suppliers:
                    total_shift += quantity
        
        return total_shift
    
    def run_scenario_batch(self, digital_twin, scenarios):
        """Run multiple scenarios and return results"""
        results = []
        
        for scenario_config in scenarios:
            scenario_type = scenario_config['type']
            params = scenario_config['params']
            
            if scenario_type in self.scenario_library:
                result = self.scenario_library[scenario_type](digital_twin, **params)
                results.append(result)
        
        return results

print("Scenario engine initialized...")
```

```python
try:
    # Run continuous simulation
    print("\n=== CONTINUOUS SIMULATION ===")
    print("\nRunning 30-day simulation with real-time optimization...")
    
    simulation_days = 30
    hours_per_step = 6  # 6-hour time steps
    total_steps = (simulation_days * 24) // hours_per_step
    
    print(f"Simulation period: {simulation_days} days")
    print(f"Time steps: {total_steps} ({hours_per_step}-hour intervals)")
    print(f"Simulation speed: {digital_twin.simulation_speed}x real-time")
    
    # Track simulation metrics
    simulation_kpis = []
    optimization_updates = 0
    disruption_events = 0
    
    start_time = time.time()
    
    for step in range(total_steps):
        # Update system state
        digital_twin.update_system_state(hours_per_step)
        
        # Random disruption events (5% chance per step)
        if np.random.random() < 0.05:
            disruption_events += 1
            
            # Select random supplier for disruption
            supplier_name = np.random.choice(list(digital_twin.suppliers.keys()))
            disruption_type = np.random.choice(['capacity_reduction', 'quality_issue', 'operational_issue'])
            severity = np.random.uniform(0.1, 0.3)
            
            digital_twin.suppliers[supplier_name].apply_disruption(disruption_type, severity)
            
            # Record disruption
            digital_twin.kpi_history['disruption_events'].append((digital_twin.current_time, supplier_name))
        
        # Re-optimize every 24 hours (4 steps)
        if step % 4 == 0:
            digital_twin.optimize_sourcing_allocation()
            optimization_updates += 1
        
        # Calculate KPIs
        kpis = digital_twin.calculate_kpis()
        simulation_kpis.append(kpis)
        
        # Progress reporting
        if step % 20 == 0 or step == total_steps - 1:
            elapsed = time.time() - start_time
            progress = (step + 1) / total_steps * 100
            print(f"Step {step+1:3d}/{total_steps} ({progress:.1f}%) - Time: {elapsed:.1f}s - KPIs updated")
    
    simulation_time = time.time() - start_time
    
    print(f"\n=== SIMULATION RESULTS ===")
    print(f"Simulation completed in {simulation_time:.2f} seconds")
    print(f"Optimization updates: {optimization_updates}")
    print(f"Disruption events: {disruption_events}")
    print(f"Average decision latency: {simulation_time / optimization_updates:.3f} seconds")
    
    # Calculate performance metrics
    if simulation_kpis:
        costs = [kpi['total_cost'] for kpi in simulation_kpis]
        risks = [kpi['supply_risk'] for kpi in simulation_kpis]
        deliveries = [kpi['delivery_performance'] for kpi in simulation_kpis]
        
        print(f"\nPerformance Metrics:")
        print(f"  Average cost: ${np.mean(costs):,.2f}")
        print(f"  Cost variability: ±${np.std(costs):,.2f}")
        print(f"  Average delivery performance: {np.mean(deliveries):.2%}")
        print(f"  Worst delivery performance: {np.min(deliveries):.2%}")
        print(f"  Risk mitigation effectiveness: {(1 - np.mean(risks)/10000):.1%}")
except Exception as e:
    print(f'Error in cell: {e}')
```
