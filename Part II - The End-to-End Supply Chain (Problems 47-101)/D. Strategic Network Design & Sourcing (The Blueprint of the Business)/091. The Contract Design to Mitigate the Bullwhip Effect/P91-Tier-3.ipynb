# 91. The Contract Design to Mitigate the Bullwhip Effect
## Tier 3: The Advanced Algorithm (Bat Algorithm for Multi-Party Contract Optimization)

### Key Assumptions
- Contract optimization landscape is complex with multiple local optima
- Bat echolocation principles can effectively balance exploration and exploitation
- Frequency modulation enables adaptive search intensity based on solution quality
- Multi-party contracts involve 25+ parameters with non-linear interactions

### Approach (Step-by-Step)
1. **Initialize bat population** with random contract parameter configurations
2. **Define echolocation parameters** (frequency, loudness, pulse rate) for adaptive search
3. **Implement frequency modulation** to balance exploration vs exploitation
4. **Apply local search** around best solutions with adaptive intensity
5. **Update population** based on solution quality and convergence criteria

### What to Look for in the Results
- Population convergence behavior and diversity metrics
- Optimal contract parameters for complex multi-party scenarios
- Bullwhip reduction achieved vs baseline and Tier 2 methods
- Algorithm performance with different population sizes and parameter settings

### Concrete Example (from the source)
5-tier electronics supply chain with 25 contract parameters:
- Bat Algorithm with 40 bats converges after 312 iterations
- Bullwhip reduction from 4.1x to 1.6x amplification
- Implementation cost under $2.3M annually
- Asymmetric information sharing (85% downstream, 65% upstream)

```python
# Import required libraries for Bat Algorithm implementation
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import pandas as pd
import time
from copy import deepcopy
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(91)

print("Libraries imported successfully for Bat Algorithm implementation")
```

```python
class BatAlgorithmContractOptimizer:
    """
    Bat Algorithm for multi-party contract optimization
    Inspired by bat echolocation behavior for global search
    """
    
    def __init__(self, n_tiers=5, n_bats=40, max_iterations=500):
        # Supply chain structure (from concrete example)
        self.n_tiers = n_tiers
        self.n_bats = n_bats
        self.max_iterations = max_iterations
        
        # Contract parameters (25 from concrete example)
        self.n_parameters = 25
        
        # Bat algorithm parameters
        self.frequency_min = 0.0
        self.frequency_max = 2.0
        self.initial_loudness = 0.5
        self.initial_pulse_rate = 0.5
        self.loudness_decay = 0.95
        self.pulse_rate_increase = 0.05
        
        # Search space bounds
        self.parameter_bounds = {
            'info_sharing': (0.1, 1.0),      # Information sharing coefficients
            'cost_allocation': (0.0, 1.0),    # Cost allocation ratios
            'penalty_structures': (0.05, 0.5), # Penalty parameters
            'capacity_terms': (0.0, 1.0),     # Capacity reservation terms
            'timing_parameters': (0.1, 0.9)   # Timing and frequency parameters
        }
        
        # Supply chain cost parameters
        self.holding_costs = [2.5, 2.0, 1.8, 1.5, 1.2]  # Decreases upstream
        self.stockout_costs = [20.0, 16.0, 14.0, 12.0, 10.0]  # Decreases upstream
        self.procurement_costs = [10.0, 8.0, 7.0, 6.0, 5.0]  # Decreases upstream
        
        # Demand characteristics
        self.base_demand = 2000
        self.demand_variance = 0.4  # 40% variance (complex scenario)
        
        # Initialize bat population
        self.bats = None
        self.frequencies = None
        self.loudness = None
        self.pulse_rates = None
        
        # Results tracking
        self.best_solution = None
        self.best_fitness = float('inf')
        self.convergence_history = []
        self.diversity_history = []
        
    def _initialize_bat_population(self):
        """
        Initialize bat population with random contract parameters
        """
        self.bats = np.zeros((self.n_bats, self.n_parameters))
        self.frequencies = np.zeros(self.n_bats)
        self.loudness = np.full(self.n_bats, self.initial_loudness)
        self.pulse_rates = np.full(self.n_bats, self.initial_pulse_rate)
        
        # Initialize each bat with random parameters
        for i in range(self.n_bats):
            self.bats[i] = self._generate_random_contract_params()
            self.frequencies[i] = np.random.uniform(self.frequency_min, self.frequency_max)
        
        # Initialize best solution
        for i in range(self.n_bats):
            fitness = self._evaluate_fitness(self.bats[i])
            if fitness < self.best_fitness:
                self.best_fitness = fitness
                self.best_solution = self.bats[i].copy()
    
    def _generate_random_contract_params(self):
        """
        Generate random contract parameters within bounds
        """
        params = np.zeros(self.n_parameters)
        
        # Information sharing coefficients (10 parameters)
        for i in range(10):
            params[i] = np.random.uniform(0.1, 1.0)
        
        # Cost allocation ratios (5 parameters)
        for i in range(5):
            params[10 + i] = np.random.uniform(0.0, 1.0)
        
        # Penalty structures (5 parameters)
        for i in range(5):
            params[15 + i] = np.random.uniform(0.05, 0.5)
        
        # Capacity reservation terms (3 parameters)
        for i in range(3):
            params[20 + i] = np.random.uniform(0.0, 1.0)
        
        # Timing parameters (2 parameters)
        for i in range(2):
            params[23 + i] = np.random.uniform(0.1, 0.9)
        
        return params
    
    def _evaluate_fitness(self, params):
        """
        Evaluate fitness of contract parameters
        Lower fitness is better (multi-objective optimization)
        """
        # Check parameter bounds
        if not self._check_parameter_bounds(params):
            return float('inf')
        
        # Calculate bullwhip amplification
        amplification = self._calculate_bullwhip_amplification(params)
        
        # Calculate implementation cost
        impl_cost = self._calculate_implementation_cost(params)
        
        # Multi-objective fitness function
        # Target: amplification < 2.0x, cost < $2.3M (from concrete example)
        amplification_penalty = max(0, amplification - 2.0) * 10
        cost_penalty = max(0, impl_cost - 2300000) / 100000  # Normalize by $100k
        
        # Combined fitness (lower is better)
        fitness = amplification + amplification_penalty + cost_penalty
        
        return fitness
    
    def _check_parameter_bounds(self, params):
        """
        Check if parameters are within valid bounds
        """
        # Information sharing bounds
        if np.any(params[:10] < 0.1) or np.any(params[:10] > 1.0):
            return False
        
        # Cost allocation bounds
        if np.any(params[10:15] < 0.0) or np.any(params[10:15] > 1.0):
            return False
        
        # Penalty structure bounds
        if np.any(params[15:20] < 0.05) or np.any(params[15:20] > 0.5):
            return False
        
        # Capacity terms bounds
        if np.any(params[20:23] < 0.0) or np.any(params[20:23] > 1.0):
            return False
        
        # Timing parameters bounds
        if np.any(params[23:25] < 0.1) or np.any(params[23:25] > 0.9):
            return False
        
        return True
    
    def _calculate_bullwhip_amplification(self, params):
        """
        Calculate bullwhip amplification for complex multi-party contracts
        Enhanced model considering asymmetric information sharing
        """
        # Extract key parameters
        info_sharing = params[:10]  # Multiple information sharing paths
        cost_allocation = params[10:15]
        penalty_structures = params[15:20]
        capacity_terms = params[20:23]
        timing_params = params[23:25]
        
        # Base amplification for 5-tier supply chain (4.1x from example)
        base_amplification = 4.1
        
        # Information sharing effectiveness (asymmetric)
        downstream_sharing = np.mean(info_sharing[:5])  # Downstream shares more
        upstream_sharing = np.mean(info_sharing[5:])   # Upstream shares less
        
        # Weighted information effectiveness (downstream has higher impact)
        info_effectiveness = (0.7 * downstream_sharing + 0.3 * upstream_sharing) ** 0.8
        
        # Cost allocation effectiveness
        cost_effectiveness = np.mean(cost_allocation) ** 0.6
        
        # Penalty structure effectiveness
        penalty_effectiveness = np.mean(penalty_structures) ** 0.7
        
        # Capacity reservation benefits
        capacity_benefits = np.mean(capacity_terms) ** 0.5
        
        # Timing optimization benefits
        timing_benefits = np.mean(timing_params) ** 0.4
        
        # Combined effectiveness
        total_effectiveness = (
            0.4 * info_effectiveness +
            0.2 * cost_effectiveness +
            0.15 * penalty_effectiveness +
            0.15 * capacity_benefits +
            0.1 * timing_benefits
        )
        
        # Calculate final amplification
        reduction_factor = 0.7 * total_effectiveness  # Max 70% reduction
        final_amplification = base_amplification * (1 - reduction_factor)
        
        return max(final_amplification, 1.0)  # Minimum 1.0x amplification
    
    def _calculate_implementation_cost(self, params):
        """
        Calculate implementation cost for complex multi-party contracts
        """
        # Information system costs
        info_costs = np.sum(params[:10] ** 1.5) * 80000  # $80k base per coefficient
        
        # Cost allocation implementation costs
        allocation_costs = np.sum(params[10:15]) * 120000  # $120k per allocation unit
        
        # Penalty system costs
        penalty_costs = np.sum(params[15:20] ** 1.2) * 100000  # $100k base per penalty
        
        # Capacity reservation costs
        capacity_costs = np.sum(params[20:23]) * 150000  # $150k per capacity unit
        
        # Timing system costs
        timing_costs = np.sum(params[23:25]) * 80000  # $80k per timing parameter
        
        total_cost = info_costs + allocation_costs + penalty_costs + capacity_costs + timing_costs
        return total_cost
    
    def _update_bat_position(self, bat_idx, iteration):
        """
        Update bat position using frequency modulation
        """
        # Update frequency
        self.frequencies[bat_idx] = self.frequency_min + 
            (self.frequency_max - self.frequency_min) * np.random.random()
        
        # Generate new position using frequency modulation
        new_position = self.bats[bat_idx].copy()
        
        # Frequency-based movement (exploration vs exploitation balance)
        frequency_factor = self.frequencies[bat_idx] / self.frequency_max
        
        if frequency_factor < 0.5:  # Exploration phase
            # Global search with larger steps
            step_size = 0.3 * frequency_factor
            for i in range(self.n_parameters):
                if np.random.random() < 0.3:  # 30% of parameters updated
                    perturbation = np.random.uniform(-step_size, step_size)
                    new_position[i] += perturbation
        else:  # Exploitation phase
            # Local search around best solution
            step_size = 0.1 * (1 - frequency_factor)
            for i in range(self.n_parameters):
                if np.random.random() < 0.5:  # 50% of parameters updated
                    # Move toward best solution with random perturbation
                    direction = self.best_solution[i] - self.bats[bat_idx][i]
                    new_position[i] += direction * step_size + np.random.uniform(-0.05, 0.05)
        
        # Ensure parameter bounds
        new_position = self._clip_parameters(new_position)
        
        return new_position
    
    def _clip_parameters(self, params):
        """
        Clip parameters to valid bounds
        """
        # Information sharing bounds
        params[:10] = np.clip(params[:10], 0.1, 1.0)
        
        # Cost allocation bounds
        params[10:15] = np.clip(params[10:15], 0.0, 1.0)
        
        # Penalty structure bounds
        params[15:20] = np.clip(params[15:20], 0.05, 0.5)
        
        # Capacity terms bounds
        params[20:23] = np.clip(params[20:23], 0.0, 1.0)
        
        # Timing parameters bounds
        params[23:25] = np.clip(params[23:25], 0.1, 0.9)
        
        return params
    
    def _local_search(self, position):
        """
        Perform local search around a promising solution
        """
        # Generate local search candidates
        candidates = []
        for _ in range(5):  # Generate 5 local candidates
            candidate = position.copy()
            
            # Small random perturbations
            for i in range(self.n_parameters):
                if np.random.random() < 0.2:  # 20% of parameters
                    perturbation = np.random.uniform(-0.02, 0.02)
                    candidate[i] += perturbation
            
            candidate = self._clip_parameters(candidate)
            candidates.append(candidate)
        
        # Find best candidate
        best_candidate = position
        best_fitness = self._evaluate_fitness(position)
        
        for candidate in candidates:
            fitness = self._evaluate_fitness(candidate)
            if fitness < best_fitness:
                best_fitness = fitness
                best_candidate = candidate
        
        return best_candidate
    
    def _calculate_population_diversity(self):
        """
        Calculate population diversity for convergence monitoring
        """
        if self.n_bats <= 1:
            return 0.0
        
        # Calculate average pairwise distance
        total_distance = 0.0
        count = 0
        
        for i in range(self.n_bats):
            for j in range(i + 1, self.n_bats):
                distance = np.linalg.norm(self.bats[i] - self.bats[j])
                total_distance += distance
                count += 1
        
        if count > 0:
            return total_distance / count
        else:
            return 0.0
    
    def optimize(self):
        """
        Main Bat Algorithm optimization loop
        """
        print("Starting Bat Algorithm for multi-party contract optimization...")
        print(f"Population size: {self.n_bats} bats")
        print(f"Problem dimension: {self.n_parameters} parameters")
        print(f"Maximum iterations: {self.max_iterations}")
        print()
        
        # Initialize population
        self._initialize_bat_population()
        
        start_time = time.time()
        
        for iteration in range(self.max_iterations):
            # Update each bat
            for i in range(self.n_bats):
                # Generate new position
                new_position = self._update_bat_position(i, iteration)
                new_fitness = self._evaluate_fitness(new_position)
                
                # Random walk with loudness and pulse rate
                if np.random.random() > self.pulse_rates[i]:
                    # Generate local solution around best
                    if np.random.random() < self.loudness[i]:
                        local_solution = self._local_search(self.best_solution)
                        local_fitness = self._evaluate_fitness(local_solution)
                        
                        # Accept local solution if better
                        if local_fitness < new_fitness:
                            new_position = local_solution
                            new_fitness = local_fitness
                
                # Accept new solution if better
                if new_fitness < self._evaluate_fitness(self.bats[i]):
                    self.bats[i] = new_position
                    
                    # Update pulse rate (increase)
                    self.pulse_rates[i] = min(1.0, 
                        self.pulse_rates[i] + self.pulse_rate_increase)
                    
                    # Update loudness (decrease)
                    self.loudness[i] *= self.loudness_decay
                    
                    # Update best solution
                    if new_fitness < self.best_fitness:
                        self.best_fitness = new_fitness
                        self.best_solution = new_position.copy()
            
            # Track convergence and diversity
            self.convergence_history.append(self.best_fitness)
            self.diversity_history.append(self._calculate_population_diversity())
            
            # Progress reporting
            if (iteration + 1) % 50 == 0:
                diversity = self._calculate_population_diversity()
                print(f"Iteration {iteration + 1}: Fitness = {self.best_fitness:.6f}, Diversity = {diversity:.4f}")
            
            # Early stopping if converged
            if iteration > 100 and len(self.convergence_history) > 50:
                recent_improvement = abs(self.convergence_history[-1] - self.convergence_history[-50])
                if recent_improvement < 1e-6:
                    print(f"Converged after {iteration + 1} iterations")
                    break
        
        optimization_time = time.time() - start_time
        print(f"\nOptimization completed in {optimization_time:.2f} seconds")
        print(f"Best fitness: {self.best_fitness:.6f}")
        
        return self.best_solution, self.best_fitness
    
    def analyze_results(self, optimal_params):
        """
        Analyze and interpret the Bat Algorithm results
        """
        print("\n=== BAT ALGORITHM OPTIMIZATION RESULTS ===")
        print()
        
        # Extract parameter categories
        info_sharing = optimal_params[:10]
        cost_allocation = optimal_params[10:15]
        penalty_structures = optimal_params[15:20]
        capacity_terms = optimal_params[20:23]
        timing_params = optimal_params[23:25]
        
        # Calculate performance metrics
        final_amplification = self._calculate_bullwhip_amplification(optimal_params)
        impl_cost = self._calculate_implementation_cost(optimal_params)
        baseline_amplification = 4.1
        reduction_percentage = ((baseline_amplification - final_amplification) / baseline_amplification) * 100
        
        print("MULTI-PARTY CONTRACT OPTIMIZATION RESULTS:")
        print(f"  Baseline amplification: {baseline_amplification:.2f}x")
        print(f"  Optimized amplification: {final_amplification:.2f}x")
        print(f"  Reduction percentage: {reduction_percentage:.1f}%")
        print(f"  Implementation cost: ${impl_cost:,.0f}")
        print()
        
        print("ASYMMETRIC INFORMATION SHARING ANALYSIS:")
        downstream_sharing = np.mean(info_sharing[:5])
        upstream_sharing = np.mean(info_sharing[5:])
        print(f"  Downstream sharing (avg): {downstream_sharing:.3f}")
        print(f"  Upstream sharing (avg): {upstream_sharing:.3f}")
        print(f"  Asymmetry ratio: {downstream_sharing/upstream_sharing:.2f}")
        print()
        
        print("COST ALLOCATION STRUCTURE:")
        tier_names = ['Tier 1-2', 'Tier 2-3', 'Tier 3-4', 'Tier 4-5', 'Cross-tier']
        for i, (name, allocation) in enumerate(zip(tier_names, cost_allocation)):
            print(f"  {name}: {allocation:.3f}")
        print()
        
        print("ALGORITHM PERFORMANCE METRICS:")
        convergence_iterations = len(self.convergence_history)
        initial_diversity = self.diversity_history[0] if self.diversity_history else 0
        final_diversity = self.diversity_history[-1] if self.diversity_history else 0
        
        print(f"  Convergence iterations: {convergence_iterations}")
        print(f"  Initial population diversity: {initial_diversity:.4f}")
        print(f"  Final population diversity: {final_diversity:.4f}")
        print(f"  Diversity reduction: {(1 - final_diversity/initial_diversity)*100:.1f}%")
        print()
        
        return {
            'amplification': final_amplification,
            'reduction_percentage': reduction_percentage,
            'impl_cost': impl_cost,
            'downstream_sharing': downstream_sharing,
            'upstream_sharing': upstream_sharing,
            'info_sharing': info_sharing,
            'cost_allocation': cost_allocation,
            'penalty_structures': penalty_structures,
            'capacity_terms': capacity_terms,
            'timing_params': timing_params
        }

print("BatAlgorithmContractOptimizer class defined successfully")
```

```python
# Initialize and run the Bat Algorithm optimizer
bat_optimizer = BatAlgorithmContractOptimizer(n_tiers=5, n_bats=40, max_iterations=500)

print("=== BAT ALGORITHM FOR MULTI-PARTY CONTRACT OPTIMIZATION ===")
print(f"Supply chain tiers: {bat_optimizer.n_tiers}")
print(f"Bat population: {bat_optimizer.n_bats}")
print(f"Contract parameters: {bat_optimizer.n_parameters}")
print(f"Frequency range: [{bat_optimizer.frequency_min}, {bat_optimizer.frequency_max}]")
print(f"Initial loudness: {bat_optimizer.initial_loudness}")
print(f"Initial pulse rate: {bat_optimizer.initial_pulse_rate}")
print()

# Run Bat Algorithm optimization
optimal_params, optimal_fitness = bat_optimizer.optimize()
```

- **Handles complex landscapes** with 25+ interacting parameters
- **Asymmetric information sharing** models realistic supply chain dynamics

- ✅ **Global optimization** - less likely to get stuck in local optima
- ✅ **Population diversity** maintains search exploration capability
- ✅ **Frequency modulation** adapts search strategy dynamically
- ✅ **Complex parameter handling** for 25+ dimensional problems
- ✅ **Asymmetric modeling** captures realistic contract dynamics

**Disadvantages vs Tier 2:**
- ❌ **Higher computational cost** - requires evaluating multiple solutions
- ❌ **More parameter tuning** - frequency, loudness, pulse rate settings
- ❌ **Less deterministic** - results vary between runs due to randomness
- ❌ **Complex implementation** - requires understanding of swarm intelligence

### When to Use This Tier

**Use Tier 3 when:**
- Contract optimization has **many local optima** and complex interactions
- **Global search capability** is needed for high-quality solutions
- **Asymmetric relationships** exist between supply chain partners
- **25+ parameters** need simultaneous optimization
- **Previous methods** (Tier 2) get stuck in suboptimal solutions

**Consider Tier 2 when:**
- Problem landscape is **relatively smooth** with fewer local optima
- **Computational efficiency** is more important than solution quality
- **Deterministic results** are required for reproducibility
- **Simple parameter interactions** don't require global search
