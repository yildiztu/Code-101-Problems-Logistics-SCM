# P91-Tier-7: Human-AI Symbiotic Partnership

## Problem Description

The Human-AI Symbiotic Partnership represents a revolutionary approach to contract design where human expertise and artificial intelligence collaborate to create optimal supply chain agreements. This tier recognizes that while AI can process vast amounts of data and identify complex patterns, human intuition, experience, and ethical judgment remain invaluable in contract design.

In this tier, we implement a **human-AI collaborative system** where:
- **AI Component**: Provides data-driven insights, optimization recommendations, and risk analysis
- **Human Component**: Offers domain expertise, ethical considerations, and strategic judgment
- **Symbiotic Interface**: Enables seamless collaboration and learning between human and AI

## Key Assumptions

1. **Human Expert**: Supply chain manager with 10+ years of experience
2. **AI System**: Advanced machine learning with explainable reasoning capabilities
3. **Trust Building**: Gradual development of human trust in AI recommendations
4. **Learning Loop**: Both human and AI learn from each other's expertise
5. **Decision Framework**: Structured approach to collaborative decision-making

## Approach

### Human-AI Collaboration Framework

#### 1. AI Capabilities
- **Pattern Recognition**: Identify complex demand patterns and correlations
- **Optimization**: Generate mathematically optimal contract parameters
- **Risk Assessment**: Quantify potential risks and uncertainties
- **Scenario Analysis**: Evaluate multiple contract scenarios simultaneously

#### 2. Human Expertise
- **Domain Knowledge**: Understanding of industry-specific constraints
- **Relationship Management**: Consideration of partner relationships and trust
- **Strategic Vision**: Long-term business strategy and market positioning
- **Ethical Judgment**: Fairness, sustainability, and social responsibility

#### 3. Symbiotic Interface
- **Explainable AI**: AI provides clear reasoning for recommendations
- **Feedback Mechanism**: Human can accept, modify, or reject AI suggestions
- **Trust Scoring**: Dynamic assessment of human confidence in AI
- **Learning Adaptation**: AI learns from human decisions and preferences

### Mathematical Framework

#### Trust Evolution Model

```
Trust_t = Trust_{t-1} + α × Success_t - β × Rejection_t
```

Where:
- *Trust_t* = Trust level at time *t*
- *Success_t* = Successful AI recommendations at time *t*
- *Rejection_t* = Rejected AI recommendations at time *t*
- *α*, *β* = Learning parameters

#### Collaborative Utility Function

```
U_collaborative = w_AI × U_AI + w_Human × U_Human + λ × Trust
```

Where:
- *U_AI* = AI-optimized utility (mathematical optimum)
- *U_Human* = Human-preferred utility (experience-based)
- *Trust* = Current trust level between human and AI
- *w_AI*, *w_Human*, *λ* = Weight parameters

#### Acceptance Probability Model

```
P_accept = 1 / (1 + exp(-γ × (AI_Confidence + Trust - Threshold)))
```

Where:
- *AI_Confidence* = AI's confidence in its recommendation
- *Trust* = Current human trust in AI
- *Threshold* = Human's acceptance threshold
- *γ* = Sensitivity parameter

## Expected Results

1. **Improved Contract Quality**: Better than AI-only or human-only approaches
2. **Trust Development**: Increasing human trust in AI recommendations over time
3. **Learning Enhancement**: Both human and AI improve performance through collaboration
4. **Risk Reduction**: Better risk assessment through combined expertise
5. **Adaptation**: System adapts to specific organizational context and preferences

````python
from typing import Tuple, List, Dict, Set

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass
from typing import List, Dict, Tuple, Optional
import random
import warnings
warnings.filterwarnings('ignore')

# Set style for better visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
````

````python
@dataclass
class HumanExpert:
    """Represents the human expert in the symbiotic partnership"""
    name: str
    experience_years: int
    expertise_areas: List[str]
    risk_tolerance: float  # 0 = risk-averse, 1 = risk-seeking
    relationship_focus: float  # 0 = purely analytical, 1 = relationship-focused
    initial_trust: float  # Initial trust in AI (0-1)
    
    def evaluate_contract(self, ai_params: Dict, ai_confidence: float, 
                          context: Dict) -> Tuple[Dict, float, str]:
        """Human expert evaluates AI contract recommendation"""
        # Start with AI recommendation
        human_params = ai_params.copy()
        
        # Apply human expertise modifications
        modifications = []
        
        # Relationship-based adjustments
        if self.relationship_focus > 0.5:
            # Human considers partner relationships
            if context.get('partner_history', {}).get('reliability', 0.5) < 0.7:
                # Less reliable partner -> more conservative terms
                human_params['safety_stock'] *= 1.2
                modifications.append("Increased safety stock for less reliable partner")
            
            if context.get('partner_history', {}).get('flexibility', 0.5) > 0.7:
                # Flexible partner -> can offer more flexibility
                human_params['flexibility_factor'] *= 1.1
                modifications.append("Increased flexibility for cooperative partner")
        
        # Risk tolerance adjustments
        if self.risk_tolerance < 0.3:
            # Risk-averse -> more conservative
            human_params['revenue_sharing'] *= 0.95  # More conservative revenue sharing
            modifications.append("Conservative revenue sharing due to risk aversion")
        elif self.risk_tolerance > 0.7:
            # Risk-seeking -> more aggressive
            human_params['lead_time_buffer'] *= 0.9  # Shorter lead time buffer
            modifications.append("Reduced lead time buffer for aggressive strategy")
        
        # Experience-based adjustments
        if self.experience_years > 10:
            # Experienced expert adds wisdom factor
            human_params['info_sharing_bonus'] *= 1.15  # Values information sharing
            modifications.append("Enhanced information sharing based on experience")
        
        # Calculate acceptance score
        acceptance_score = self._calculate_acceptance_score(ai_params, human_params, 
                                                        ai_confidence, context)
        
        # Generate reasoning
        reasoning = f"Human expert {self.name} (" 
        reasoning += f"{self.experience_years} years experience) "
        reasoning += f"made {len(modifications)} modifications: " + "; ".join(modifications)
        
        return human_params, acceptance_score, reasoning
    
    def _calculate_acceptance_score(self, ai_params: Dict, human_params: Dict, 
                                   ai_confidence: float, context: Dict) -> float:
        """Calculate how much the human accepts the AI recommendation"""
        # Calculate parameter differences
        param_diff = 0
        for key in ai_params:
            if key in human_params:
                diff = abs(human_params[key] - ai_params[key]) / ai_params[key]
                param_diff += diff
        
        avg_diff = param_diff / len(ai_params)
        
        # Acceptance decreases with larger differences
        acceptance = max(0, 1 - avg_diff * 2)
        
        # Adjust for AI confidence
        acceptance *= (0.5 + 0.5 * ai_confidence)
        
        return acceptance
````

````python
class ExplainableAI:
    """AI system with explainable reasoning capabilities"""
    
    def __init__(self, model_complexity: str = "advanced"):
        self.model_complexity = model_complexity
        self.confidence_threshold = 0.7
        self.learning_rate = 0.1
        self.human_preferences = {}  # Learn from human decisions
        
    def analyze_demand_patterns(self, demand_data: np.ndarray) -> Dict:
        """Analyze demand patterns and extract insights"""
        # Calculate statistical properties
        stats = {
            'mean': np.mean(demand_data),
            'std': np.std(demand_data),
            'cv': np.std(demand_data) / np.mean(demand_data),
            'trend': self._calculate_trend(demand_data),
            'seasonality': self._detect_seasonality(demand_data)
        }
        
        # Generate insights
        insights = []
        
        if stats['cv'] > 0.3:
            insights.append("High demand variability detected - recommend flexible contracts")
        
        if abs(stats['trend']) > 0.1:
            insights.append(f"{'Increasing' if stats['trend'] > 0 else 'Decreasing'} trend detected")
        
        if stats['seasonality'] > 0.2:
            insights.append("Seasonal patterns detected - consider seasonal contracts")
        
        return {
            'statistics': stats,
            'insights': insights,
            'complexity_score': self._assess_complexity(stats)
        }
    
    def generate_contract_recommendation(self, demand_analysis: Dict, 
                                        context: Dict) -> Tuple[Dict, float, List[str]]:
        """Generate contract recommendation with confidence and reasoning"""
        stats = demand_analysis['statistics']
        insights = demand_analysis['insights']
        
        # Base contract parameters
        params = {
            'revenue_sharing': 0.5,
            'flexibility_factor': 1.0,
            'safety_stock': 1.5,
            'info_sharing_bonus': 50,
            'lead_time_buffer': 0.5
        }
        
        reasoning = []
        
        # Adjust based on demand characteristics
        if stats['cv'] > 0.3:
            params['flexibility_factor'] = 1.2
            params['safety_stock'] = 2.0
            reasoning.append("Increased flexibility and safety stock for high variability")
        
        if stats['trend'] > 0.1:
            params['revenue_sharing'] = 0.55  # Favor growth
            reasoning.append("Adjusted revenue sharing for growth trend")
        elif stats['trend'] < -0.1:
            params['revenue_sharing'] = 0.45  # Conservative for decline
            reasoning.append("Conservative revenue sharing for declining trend")
        
        # Apply learned human preferences
        if self.human_preferences:
            for param, adjustment in self.human_preferences.items():
                if param in params:
                    params[param] *= adjustment
                    reasoning.append(f"Applied learned preference for {param}")
        
        # Calculate confidence
        confidence = self._calculate_confidence(stats, context)
        
        return params, confidence, reasoning
    
    def learn_from_human(self, human_params: Dict, ai_params: Dict, 
                        acceptance_score: float):
        """Learn from human modifications"""
        if acceptance_score < 0.5:  # Human significantly modified
            for param in ai_params:
                if param in human_params:
                    ratio = human_params[param] / ai_params[param]
                    if abs(ratio - 1.0) > 0.1:  # Significant modification
                        # Update learned preference
                        if param not in self.human_preferences:
                            self.human_preferences[param] = 1.0
                        
                        # Gradual learning
                        self.human_preferences[param] *= (1 - self.learning_rate)
                        self.human_preferences[param] += self.learning_rate * ratio
    
    def _calculate_trend(self, demand_data: np.ndarray) -> float:
        """Calculate linear trend in demand data"""
        x = np.arange(len(demand_data))
        return np.corrcoef(x, demand_data)[0, 1]
    
    def _detect_seasonality(self, demand_data: np.ndarray) -> float:
        """Detect seasonal patterns in demand data"""
        if len(demand_data) < 12:
            return 0.0
        
        # Simple seasonality detection using autocorrelation
        seasonal_corr = []
        for lag in [4, 8, 12]:  # Quarterly, 2-monthly, yearly patterns
            if len(demand_data) > lag:
                corr = np.corrcoef(demand_data[:-lag], demand_data[lag:])[0, 1]
                seasonal_corr.append(abs(corr))
        
        return np.mean(seasonal_corr) if seasonal_corr else 0.0
    
    def _assess_complexity(self, stats: Dict) -> float:
        """Assess complexity of demand patterns"""
        complexity = 0.0
        
        # High variability increases complexity
        complexity += min(stats['cv'], 1.0) * 0.4
        
        # Strong trend increases complexity
        complexity += min(abs(stats['trend']), 1.0) * 0.3
        
        # Seasonality increases complexity
        complexity += min(stats['seasonality'], 1.0) * 0.3
        
        return complexity
    
    def _calculate_confidence(self, stats: Dict, context: Dict) -> float:
        """Calculate AI confidence in recommendations"""
        base_confidence = 0.8
        
        # Reduce confidence for high complexity
        complexity = self._assess_complexity(stats)
        confidence = base_confidence * (1 - complexity * 0.3)
        
        # Increase confidence with more data
        data_quality = context.get('data_quality', 0.8)
        confidence *= data_quality
        
        # Adjust for model complexity
        if self.model_complexity == "advanced":
            confidence *= 1.1
        
        return min(0.95, max(0.5, confidence))
````

````python
class SymbioticPartnership:
    """Manages the human-AI symbiotic partnership"""
    
    def __init__(self, human_expert: HumanExpert, ai_system: ExplainableAI):
        self.human = human_expert
        self.ai = ai_system
        self.trust_level = human_expert.initial_trust
        self.session_history = []
        self.performance_metrics = []
        
        # Trust evolution parameters
        self.trust_alpha = 0.1  # Learning rate for positive experiences
        self.trust_beta = 0.05  # Learning rate for negative experiences
        
    def conduct_collaborative_session(self, demand_data: np.ndarray, 
                                      context: Dict) -> Dict:
        """Conduct one collaborative contract design session"""
        session_result = {
            'session_id': len(self.session_history) + 1,
            'demand_analysis': None,
            'ai_recommendation': None,
            'human_evaluation': None,
            'final_contract': None,
            'trust_before': self.trust_level,
            'trust_after': None,
            'performance': None
        }
        
        # Step 1: AI analyzes demand patterns
        demand_analysis = self.ai.analyze_demand_patterns(demand_data)
        session_result['demand_analysis'] = demand_analysis
        
        # Step 2: AI generates recommendation
        ai_params, ai_confidence, ai_reasoning = self.ai.generate_contract_recommendation(
            demand_analysis, context)
        
        session_result['ai_recommendation'] = {
            'parameters': ai_params,
            'confidence': ai_confidence,
            'reasoning': ai_reasoning
        }
        
        # Step 3: Human evaluates and modifies
        human_params, acceptance_score, human_reasoning = self.human.evaluate_contract(
            ai_params, ai_confidence, context)
        
        session_result['human_evaluation'] = {
            'parameters': human_params,
            'acceptance_score': acceptance_score,
            'reasoning': human_reasoning
        }
        
        # Step 4: Trust Update
        self._update_trust(acceptance_score, ai_confidence)
        session_result['trust_after'] = self.trust_level
        
        # Step 5: AI Learning
        human_feedback = {
            "acceptance_score": acceptance_score,
            "ai_confidence": ai_confidence,
            "modifications": human_params,
            "final_parameters": human_params
        }
        
        self.ai.learn_from_human(human_params, ai_params, acceptance_score)
        
        # Step 6: Final contract and performance
        final_contract = human_params  # Human has final say
        session_result['final_contract'] = final_contract
        
        # Simulate contract performance
        performance = self._simulate_contract_performance(final_contract, demand_data)
        session_result['performance'] = performance
        
        # Store session
        self.session_history.append(session_result)
        self.performance_metrics.append(performance)
        
        return session_result
    
    def _update_trust(self, acceptance_score: float, ai_confidence: float):
        """Update trust level based on interaction"""
        # Positive reinforcement when AI is confident and human accepts
        if ai_confidence > 0.7 and acceptance_score > 0.7:
            self.trust_level += self.trust_alpha * (1 - self.trust_level)
        
        # Negative reinforcement when AI is overconfident but human rejects
        elif ai_confidence > 0.7 and acceptance_score < 0.3:
            self.trust_level -= self.trust_beta * self.trust_level
        
        # Keep trust in reasonable bounds
        self.trust_level = max(0.1, min(0.95, self.trust_level))
    
    def _simulate_contract_performance(self, contract_params: Dict, 
                                       demand_data: np.ndarray) -> Dict:
        """Simulate contract performance with bullwhip effect"""
        # Simplified performance simulation
        demand_std = np.std(demand_data)
        demand_mean = np.mean(demand_data)
        
        # Calculate order variability based on contract
        flexibility = contract_params.get('flexibility_factor', 1.0)
        safety_stock = contract_params.get('safety_stock', 1.5)
        
        # Order amplification factor
        order_amplification = (2.0 - flexibility) * (safety_stock / 1.5)
        order_std = demand_std * order_amplification
        
        # Bullwhip ratio
        bullwhip_ratio = order_std / (demand_std + 1e-6)
        
        # Calculate costs
        holding_cost = demand_mean * safety_stock * 0.2
        stockout_cost = demand_std * order_amplification * 0.1
        
        # Revenue (simplified)
        revenue = demand_mean * 100 * contract_params.get('revenue_sharing', 0.5)
        
        # Information sharing benefit
        info_benefit = contract_params.get('info_sharing_bonus', 0) * 0.5
        
        total_cost = holding_cost + stockout_cost
        net_profit = revenue + info_benefit - total_cost
        
        return {
            'bullwhip_ratio': bullwhip_ratio,
            'total_cost': total_cost,
            'net_profit': net_profit,
            'service_level': max(0, 1 - stockout_cost / (demand_mean * 10))
        }
    
    def get_partnership_summary(self) -> Dict:
        """Get summary of partnership performance"""
        if not self.session_history:
            return {"message": "No sessions conducted yet"}
        
        # Calculate metrics
        trust_evolution = [s['trust_before'] for s in self.session_history]
        trust_evolution.append(self.trust_level)
        
        acceptance_scores = [s['human_evaluation']['acceptance_score'] 
                           for s in self.session_history]
        
        avg_bullwhip = np.mean([p['bullwhip_ratio'] for p in self.performance_metrics])
        avg_profit = np.mean([p['net_profit'] for p in self.performance_metrics])
        
        return {
            'total_sessions': len(self.session_history),
            'trust_evolution': trust_evolution,
            'final_trust_level': self.trust_level,
            'average_acceptance': np.mean(acceptance_scores),
            'average_bullwhip_ratio': avg_bullwhip,
            'average_profit': avg_profit,
            'trust_improvement': self.trust_level - self.human.initial_trust
        }
````

````python
# Initialize the symbiotic partnership
human_expert = HumanExpert(
    name="Sarah Chen",
    experience_years=12,
    expertise_areas=["contract_design", "relationship_management", "risk_assessment"],
    risk_tolerance=0.4,  # Moderately risk-averse
    relationship_focus=0.7,  # Values relationships
    initial_trust=0.5  # Neutral initial trust
)

ai_system = ExplainableAI(model_complexity="advanced")

partnership = SymbioticPartnership(human_expert, ai_system)

print("=== Human-AI Symbiotic Partnership Initialized ===")
print(f"Human Expert: {human_expert.name} ({human_expert.experience_years} years experience)")
print(f"Initial Trust Level: {partnership.trust_level:.2f}")
print(f"AI Model Complexity: {ai_system.model_complexity}")
````

````python
try:
    # Generate demand data for simulation
    def generate_complex_demand(periods: int) -> np.ndarray:
        """Generate complex demand pattern with trend, seasonality, and noise"""
        np.random.seed(42)
        
        demand = np.zeros(periods)
        base_demand = 1000
        
        for t in range(periods):
            # Trend component
            trend = 5 * t / periods  # Slight upward trend
            
            # Seasonal component
            seasonal = 200 * np.sin(2 * np.pi * t / 12)  # Yearly seasonality
            
            # Random noise
            noise = np.random.normal(0, 150)
            
            demand[t] = max(100, base_demand + trend + seasonal + noise)
        
        return demand
    
    # Create demand scenarios for different sessions
    scenarios = {
        'stable': generate_complex_demand(52),  # One year
        'volatile': generate_complex_demand(52) * np.random.uniform(0.8, 1.2, 52),
        'growing': generate_complex_demand(52) * np.linspace(1.0, 1.3, 52),
        'declining': generate_complex_demand(52) * np.linspace(1.0, 0.8, 52)
    }
    
    print(f"Generated {len(scenarios)} demand scenarios")
    for name, demand in scenarios.items():
        print(f"{name}: Mean={np.mean(demand):.1f}, Std={np.std(demand):.1f}")
except Exception as e:
    print(f'Error in cell: {e}')
````

````python
try:
    # Run collaborative sessions
    print("=== Running Collaborative Sessions ===")
    
    context = {
        'partner_history': {
            'reliability': 0.75,
            'flexibility': 0.65,
            'communication_quality': 0.8
        },
        'data_quality': 0.85,
        'market_conditions': 'moderate'
    }
    
    session_results = []
    
    for i, (scenario_name, demand_data) in enumerate(scenarios.items()):
        print(f"\n--- Session {i+1}: {scenario_name.title()} Scenario ---")
        
        # Update context for scenario
        scenario_context = context.copy()
        scenario_context['scenario_name'] = scenario_name
        
        # Conduct collaborative session
        session_result = partnership.conduct_collaborative_session(demand_data, scenario_context)
        session_results.append(session_result)
        
        # Print session summary
        ai_rec = session_result['ai_recommendation']
        human_eval = session_result['human_evaluation']
        performance = session_result['performance']
        
        print(f"AI Confidence: {ai_rec['confidence']:.3f}")
        print(f"Human Acceptance: {human_eval['acceptance_score']:.3f}")
        print(f"Trust Level: {session_result['trust_before']:.3f} → {session_result['trust_after']:.3f}")
        print(f"Bullwhip Ratio: {performance['bullwhip_ratio']:.3f}")
        print(f"Net Profit: ${performance['net_profit']:,.2f}")
except Exception as e:
    print(f'Error in cell: {e}')
````
