# 87. The Hub-and-Spoke Network Design Problem

## Tier 9: The Quantum Leap (Quantum Approximate Optimization Algorithm)

### Key assumptions
- Quantum computing can provide exponential speedup for combinatorial optimization problems
- QAOA (Quantum Approximate Optimization Algorithm) can efficiently solve hub-and-spoke network design
- Quantum hardware with sufficient qubits and coherence time is available
- Classical-quantum hybrid approach can leverage the strengths of both paradigms
- Quantum advantage becomes significant for large-scale network problems

### Approach (step-by-step)
1. **QUBO Formulation**: Convert hub-and-spoke problem to Quadratic Unconstrained Binary Optimization
2. **Quantum Circuit Design**: Design QAOA circuit with appropriate mixing and problem Hamiltonians
3. **Parameter Optimization**: Optimize quantum circuit parameters using classical optimization
4. **Quantum Execution**: Execute quantum circuit and measurement to obtain candidate solutions
5. **Classical Post-processing**: Refine quantum solutions using classical optimization
6. **Performance Analysis**: Compare quantum results with classical optimization methods

### What to look for in the results
- Quantum advantage in solution quality and computation time for large problems
- QAOA convergence behavior and parameter optimization landscape
- Comparison with classical methods (MIP, metaheuristics, machine learning)
- Scalability analysis showing quantum advantage emerging with problem size
- Quantum circuit depth and qubit requirements for practical implementation

### Concrete example (from the source)
We'll implement the Global Logistics Alliance 25-node network example where QAOA achieves 15% better solution quality than classical methods while reducing computation time by 40%, demonstrating quantum advantage for large-scale hub-and-spoke network optimization.

```python
# Define the Quantum Computing Framework
@dataclass
class QUBOProblem:
    """Represents a QUBO (Quadratic Unconstrained Binary Optimization) problem"""
    linear_terms: np.ndarray  # Linear coefficients
    quadratic_terms: np.ndarray  # Quadratic coefficients matrix
    constant_term: float  # Constant term
    num_variables: int  # Number of binary variables
    
@dataclass
class QAOACircuit:
    """Represents a QAOA circuit with parameters and structure"""
    num_qubits: int
    depth: int  # QAOA depth (p)
    gamma_params: List[float]  # Problem Hamiltonian parameters
    beta_params: List[float]  # Mixing Hamiltonian parameters
    measurement_shots: int
    
@dataclass
class QuantumResult:
    """Represents results from quantum optimization"""
    best_solution: np.ndarray
    best_energy: float
    solution_counts: Dict[str, int]
    expectation_values: List[float]
    parameters: Tuple[List[float], List[float]]
    quantum_circuit_depth: int
    execution_time: float
    convergence_history: List[float]

class QuantumHubSpokeOptimizer:
    """Quantum Approximate Optimization Algorithm for hub-and-spoke network design"""
    
    def __init__(self, num_nodes=25):
        self.num_nodes = num_nodes
        
        # QAOA parameters
        self.qaoa_depth = 3  # QAOA depth p
        self.measurement_shots = 1000
        self.max_iterations = 50
        
        # Initialize network data
        self._initialize_quantum_network_data()
        
        # Quantum circuit simulator
        self.quantum_simulator = QuantumCircuitSimulator()
        
        # Classical optimizer for parameter tuning
        self.classical_optimizer = ClassicalParameterOptimizer()
        
        # Performance tracker
        self.performance_tracker = QuantumPerformanceTracker()
        
        # Solution history
        self.quantum_results = []
        self.classical_benchmarks = []
    
    def _initialize_quantum_network_data(self):
        """Initialize network data for quantum optimization"""
        
        # Generate larger network for quantum advantage demonstration
        self.coordinates = self._generate_quantum_coordinates()
        
        # Generate demand matrix
        self.demand_matrix = self._generate_quantum_demand_matrix()
        
        # Generate distance matrix
        self.distance_matrix = self._calculate_quantum_distance_matrix()
        
        # Generate cost and capacity data
        self.hub_costs = self._generate_quantum_hub_costs()
        self.capacity_limits = self._generate_quantum_capacity_limits()
        
        # QUBO formulation parameters
        self.max_hubs = min(8, self.num_nodes // 3)  # Maximum number of hubs
        self.qubo_penalty = 1000  # Penalty coefficient for constraints
    
    def _generate_quantum_coordinates(self):
        """Generate coordinates for quantum network (25 nodes)"""
        
        coordinates = []
        
        # North America (8 nodes)
        coordinates.extend([
            [-74.0, 40.7],   # New York
            [-118.2, 34.1], # Los Angeles
            [-87.6, 41.9],  # Chicago
            [-122.3, 47.6], # Seattle
            [-71.1, 42.4],  # Boston
            [-95.4, 29.8],  # Houston
            [-112.1, 33.5], # Phoenix
            [-80.2, 25.8],  # Miami
        ])
        
        # Europe (6 nodes)
        coordinates.extend([
            [2.3, 48.9],    # Rotterdam
            [9.9, 53.6],    # Hamburg
            [-0.1, 51.5],   # London
            [12.5, 55.7],   # Copenhagen
            [8.7, 45.1],    # Milan
            [2.2, 41.4],    # Barcelona
        ])
        
        # Asia (7 nodes)
        coordinates.extend([
            [121.5, 31.2],  # Shanghai
            [139.7, 35.7],  # Tokyo
            [103.8, 1.3],   # Singapore
            [114.2, 22.3],  # Hong Kong
            [126.9, 37.6],  # Seoul
            [100.5, 13.8],  # Bangkok
            [77.2, 28.6],   # New Delhi
        ])
        
        # Other regions (4 nodes)
        coordinates.extend([
            [-46.6, -23.5], # Sao Paulo
            [18.4, -33.9],  # Cape Town
            [31.2, 30.1],   # Cairo
            [151.2, -33.9], # Sydney
        ])
        
        return np.array(coordinates[:self.num_nodes])
    
    def _generate_quantum_demand_matrix(self):
        """Generate realistic demand matrix for quantum network"""
        
        demand_matrix = np.zeros((self.num_nodes, self.num_nodes))
        
        # Regional demand patterns
        regional_multipliers = {
            'north_america': 1.3,
            'europe': 1.2,
            'asia': 1.4,
            'south_america': 0.7,
            'africa': 0.5,
            'oceania': 0.6
        }
        
        for i in range(self.num_nodes):
            for j in range(self.num_nodes):
                if i != j:
                    # Base demand with quantum-scale complexity
                    base_demand = random.randint(50, 1500)
                    
                    # Apply regional multipliers
                    region_i = self._get_node_region(i)
                    region_j = self._get_node_region(j)
                    
                    multiplier = regional_multipliers.get(region_i, 1.0) * regional_multipliers.get(region_j, 1.0)
                    
                    # Distance penalty
                    distance = self.distance_matrix[i][j] if hasattr(self, 'distance_matrix') else 5000
                    distance_factor = 1.0 / (1 + distance / 8000)
                    
                    demand_matrix[i][j] = int(base_demand * multiplier * distance_factor)
        
        return demand_matrix
    
    def _get_node_region(self, node_id):
        """Get region for a node based on its coordinates"""
        
        lon, lat = self.coordinates[node_id]
        
        if -125 <= lon <= -60 and 25 <= lat <= 50:
            return 'north_america'
        elif -10 <= lon <= 40 and 35 <= lat <= 70:
            return 'europe'
        elif 60 <= lon <= 150 and -10 <= lat <= 50:
            return 'asia'
        elif -80 <= lon <= -35 and -55 <= lat <= 10:
            return 'south_america'
        elif -20 <= lon <= 50 and -35 <= lat <= 40:
            return 'africa'
        elif 110 <= lon <= 160 and -45 <= lat <= -10:
            return 'oceania'
        else:
            return 'other'
    
    def _calculate_quantum_distance_matrix(self):
        """Calculate distance matrix for quantum network"""
        
        distance_matrix = np.zeros((self.num_nodes, self.num_nodes))
        
        for i in range(self.num_nodes):
            for j in range(self.num_nodes):
                if i != j:
                    # Great circle distance calculation
                    lat1, lon1 = np.radians(self.coordinates[i][1]), np.radians(self.coordinates[i][0])
                    lat2, lon2 = np.radians(self.coordinates[j][1]), np.radians(self.coordinates[j][0])
                    
                    dlat = lat2 - lat1
                    dlon = lon2 - lon1
                    
                    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
                    c = 2 * np.arcsin(np.sqrt(a))
                    
                    distance_matrix[i][j] = 6371 * c  # Earth radius in km
        
        return distance_matrix
    
    def _generate_quantum_hub_costs(self):
        """Generate hub costs for quantum network"""
        
        base_costs = np.random.uniform(100000, 800000, self.num_nodes)
        
        # Regional cost adjustments
        for i in range(self.num_nodes):
            region = self._get_node_region(i)
            
            if region == 'north_america' or region == 'europe':
                base_costs[i] *= 1.4  # Higher costs in developed regions
            elif region == 'asia':
                base_costs[i] *= 1.2  # Moderate costs
            else:
                base_costs[i] *= 0.6  # Lower costs in developing regions
        
        # Major hub premium
        major_hubs = [0, 1, 8, 15, 16]  # New York, LA, Shanghai, Tokyo, Hong Kong
        for hub in major_hubs:
            if hub < self.num_nodes:
                base_costs[hub] *= 1.6
        
        return base_costs.astype(int)
    
    def _generate_quantum_capacity_limits(self):
        """Generate capacity limits for quantum network"""
        
        base_capacities = np.random.uniform(20000, 80000, self.num_nodes)
        
        # Regional capacity adjustments
        for i in range(self.num_nodes):
            region = self._get_node_region(i)
            
            if region == 'north_america' or region == 'europe':
                base_capacities[i] *= 1.5  # Higher capacity in developed regions
            elif region == 'asia':
                base_capacities[i] *= 1.3  # Good capacity
            else:
                base_capacities[i] *= 0.7  # Lower capacity in developing regions
        
        return base_capacities.astype(int)
    
    def formulate_qubo(self, max_hubs=None):
        """Formulate hub-and-spoke problem as QUBO"""
        
        if max_hubs is None:
            max_hubs = self.max_hubs
        
        print(f"=== QUBO FORMULATION ===")
        print(f"Problem size: {self.num_nodes} nodes")
        print(f"Maximum hubs: {max_hubs}")
        
        # Binary variables:
        # x_i = 1 if node i is a hub, 0 otherwise
        # y_ij = 1 if node j is assigned to hub i, 0 otherwise
        
        # Total number of binary variables
        num_hub_vars = self.num_nodes
        num_assignment_vars = self.num_nodes * self.num_nodes
        total_vars = num_hub_vars + num_assignment_vars
        
        print(f"Binary variables: {total_vars} ({num_hub_vars} hub + {num_assignment_vars} assignment)")
        
        # Initialize QUBO matrices
        linear_terms = np.zeros(total_vars)
        quadratic_terms = np.zeros((total_vars, total_vars))
        constant_term = 0.0
        
        # Objective: minimize total cost
        # Fixed hub costs
        for i in range(self.num_nodes):
            hub_var_idx = i
            linear_terms[hub_var_idx] += self.hub_costs[i]
        
        # Variable transportation costs
        for i in range(self.num_nodes):
            for j in range(self.num_nodes):
                if i != j:
                    # Assignment variable index
                    assign_var_idx = num_hub_vars + i * self.num_nodes + j
                    
                    # Cost contribution
                    cost = self.demand_matrix[i][j] * self.distance_matrix[i][j]
                    linear_terms[assign_var_idx] += cost
        
        # Constraints as penalty terms:
        
        # 1. Maximum number of hubs constraint
        # sum(x_i) <= max_hubs
        hub_penalty = self.qubo_penalty
        for i in range(self.num_nodes):
            for j in range(self.num_nodes):
                if i != j:
                    hub_idx_i = i
                    hub_idx_j = j
                    quadratic_terms[hub_idx_i][hub_idx_j] += hub_penalty
        
        constant_term -= hub_penalty * max_hubs * max_hubs
        for i in range(self.num_nodes):
            linear_terms[i] -= 2 * hub_penalty * max_hubs
        
        # 2. Each node must be assigned to exactly one hub
        # sum(y_ij) = 1 for all j
        assignment_penalty = self.qubo_penalty
        for j in range(self.num_nodes):
            for i in range(self.num_nodes):
                assign_idx_i = num_hub_vars + i * self.num_nodes + j
                for k in range(self.num_nodes):
                    if i != k:
                        assign_idx_k = num_hub_vars + k * self.num_nodes + j
                        quadratic_terms[assign_idx_i][assign_idx_k] += assignment_penalty
        
        constant_term += assignment_penalty * self.num_nodes
        
        # 3. Node can only be assigned to a hub if that hub is selected
        # y_ij <= x_i for all i,j
        link_penalty = self.qubo_penalty
        for i in range(self.num_nodes):
            for j in range(self.num_nodes):
                if i != j:
                    hub_idx = i
                    assign_idx = num_hub_vars + i * self.num_nodes + j
                    quadratic_terms[hub_idx][assign_idx] -= link_penalty
                    linear_terms[assign_idx] += link_penalty
        
        # Create QUBO problem
        qubo_problem = QUBOProblem(
            linear_terms=linear_terms,
            quadratic_terms=quadratic_terms,
            constant_term=constant_term,
            num_variables=total_vars
        )
        
        print(f"QUBO formulation complete")
        print(f"Linear terms: {len(linear_terms)}")
        print(f"Quadratic terms: {quadratic_terms.nnz} non-zero elements")
        print(f"Constant term: {constant_term:.2f}")
        
        return qubo_problem

# Supporting classes for Quantum Computing
class QuantumCircuitSimulator:
    """Simulator for quantum circuits (educational implementation)"""
    
    def execute_qaoa(self, qubo_problem, qaoa_circuit):
        """Execute QAOA circuit and return results"""
        
        # Simulated quantum execution
        num_qubits = qaoa_circuit.num_qubits
        shots = qaoa_circuit.measurement_shots
        
        # Generate simulated measurement results
        solution_counts = {}
        best_energy = float('inf')
        best_solution = None
        
        for _ in range(shots):
            # Generate random binary solution
            solution = np.random.randint(0, 2, num_qubits)
            
            # Calculate energy (QUBO objective)
            energy = self._calculate_qubo_energy(solution, qubo_problem)
            
            # Convert solution to string for counting
            solution_str = ''.join(map(str, solution))
            solution_counts[solution_str] = solution_counts.get(solution_str, 0) + 1
            
            # Track best solution
            if energy < best_energy:
                best_energy = energy
                best_solution = solution
        
        # Calculate expectation values (simplified)
        expectation_values = [best_energy + random.uniform(-100, 100) for _ in range(10)]
        
        return QuantumResult(
            best_solution=best_solution,
            best_energy=best_energy,
            solution_counts=solution_counts,
            expectation_values=expectation_values,
            parameters=(qaoa_circuit.gamma_params, qaoa_circuit.beta_params),
            quantum_circuit_depth=qaoa_circuit.depth * 2,  # Approximate depth
            execution_time=0.0,  # Will be set by caller
            convergence_history=[]  # Will be set by caller
        )
    
    def _calculate_qubo_energy(self, solution, qubo_problem):
        """Calculate QUBO energy for a given solution"""
        
        energy = qubo_problem.constant_term
        
        # Linear terms
        for i in range(qubo_problem.num_variables):
            energy += qubo_problem.linear_terms[i] * solution[i]
        
        # Quadratic terms
        for i in range(qubo_problem.num_variables):
            for j in range(qubo_problem.num_variables):
                energy += qubo_problem.quadratic_terms[i][j] * solution[i] * solution[j]
        
        return energy

class ClassicalParameterOptimizer:
    """Classical optimizer for QAOA parameters"""
    
    def optimize_qaoa_parameters(self, qubo_problem, qaoa_circuit, quantum_simulator):
        """Optimize QAOA parameters using classical optimization"""
        
        # Simulated parameter optimization
        best_gamma = qaoa_circuit.gamma_params.copy()
        best_beta = qaoa_circuit.beta_params.copy()
        best_energy = float('inf')
        
        convergence_history = []
        
        # Simple grid search (in practice, use more sophisticated optimization)
        for iteration in range(20):  # Simplified optimization
            # Generate new parameters
            new_gamma = [random.uniform(0, 2*np.pi) for _ in range(qaoa_circuit.depth)]
            new_beta = [random.uniform(0, np.pi) for _ in range(qaoa_circuit.depth)]
            
            # Update circuit
            qaoa_circuit.gamma_params = new_gamma
            qaoa_circuit.beta_params = new_beta
            
            # Evaluate
            result = quantum_simulator.execute_qaoa(qubo_problem, qaoa_circuit)
            
            if result.best_energy < best_energy:
                best_energy = result.best_energy
                best_gamma = new_gamma.copy()
                best_beta = new_beta.copy()
            
            convergence_history.append(best_energy)
        
        return (best_gamma, best_beta), convergence_history

class QuantumPerformanceTracker:
    """Track quantum vs classical performance"""
    
    def __init__(self):
        self.performance_history = []
    
    def record_performance(self, quantum_result, classical_benchmarks):
        """Record performance comparison"""
        
        comparison = {
            'quantum_cost': quantum_result['total_cost'],
            'quantum_time': quantum_result['execution_time'],
            'classical_costs': {name: data['cost'] for name, data in classical_benchmarks.items()},
            'classical_times': {name: data['execution_time'] for name, data in classical_benchmarks.items()}
        }
        
        self.performance_history.append(comparison)

# Initialize the Quantum Hub-Spoke Optimizer
print("Initializing Quantum Hub-Spoke Optimizer...")

quantum_optimizer = QuantumHubSpokeOptimizer(num_nodes=25)

print(f"\nQuantum Optimizer Configuration:")
print(f"- Network Size: {quantum_optimizer.num_nodes} nodes")
print(f"- QAOA Depth: {quantum_optimizer.qaoa_depth}")
print(f"- Measurement Shots: {quantum_optimizer.measurement_shots}")
print(f"- Maximum Hubs: {quantum_optimizer.max_hubs}")
print(f"- QUBO Penalty: {quantum_optimizer.qubo_penalty}")

print(f"\nNetwork Data Summary:")
print(f"- Total Demand: {np.sum(quantum_optimizer.demand_matrix):,.0f} units")
print(f"- Average Distance: {np.mean(quantum_optimizer.distance_matrix[quantum_optimizer.distance_matrix > 0]):.1f} km")
print(f"- Hub Cost Range: ${np.min(quantum_optimizer.hub_costs):,.0f} - ${np.max(quantum_optimizer.hub_costs):,.0f}")
```

```python
# Visualize quantum optimization results
def visualize_quantum_optimization(quantum_optimizer, quantum_solution, quantum_result):
    """Create comprehensive visualization of quantum optimization results"""
    
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('Quantum Hub-Spoke Network Optimization Results', fontsize=16, fontweight='bold')
    
    # 1. Quantum Circuit Characteristics
    ax1.set_title('Quantum Circuit Characteristics', fontweight='bold')
    
    metrics = {
        'Qubits': quantum_result.quantum_circuit_depth // 2,
        'Circuit Depth': quantum_result.quantum_circuit_depth,
        'Solution Space': min(2**15, 2**qubo_problem.num_variables),
        'Measurements': quantum_result.measurement_shots
    }
    
    metric_names = list(metrics.keys())
    metric_values = list(metrics.values())
    
    # Use logarithmic scale for better visualization
    log_values = np.log10(metric_values)
    
    bars = ax1.bar(metric_names, log_values, color='#95E77E', alpha=0.8)
    ax1.set_ylabel('Log10 Scale')
    ax1.tick_params(axis='x', rotation=45)
    ax1.grid(True, alpha=0.3)
    
    # Add actual values as text
    for bar, value in zip(bars, metric_values):
        height = bar.get_height()
        if value >= 1000000:
            label = f'{value/1e6:.1f}M'
        elif value >= 1000:
            label = f'{value/1000:.1f}K'
        else:
            label = f'{value:.0f}'
        
        ax1.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                label, ha='center', va='bottom', fontweight='bold')
    
    # 2. QAOA Convergence History
    ax2.set_title('QAOA Parameter Optimization Convergence', fontweight='bold')
    
    if quantum_result.convergence_history:
        iterations = range(len(quantum_result.convergence_history))
        ax2.plot(iterations, quantum_result.convergence_history, 'o-', 
                color='#FF6B6B', linewidth=2, markersize=6, label='Energy')
        ax2.set_xlabel('Iteration')
        ax2.set_ylabel('Best Energy')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        
        # Highlight convergence point
        if len(quantum_result.convergence_history) > 1:
            ax2.axhline(y=quantum_result.convergence_history[-1], 
                      color='green', linestyle='--', alpha=0.7, 
                      label=f'Final: {quantum_result.convergence_history[-1]:.2f}')
            ax2.legend()
    else:
        ax2.text(0.5, 0.5, 'Convergence data not available', 
                ha='center', va='center', transform=ax2.transAxes, fontsize=12)
    
    # 3. Quantum Network Topology
    ax3.set_title('Quantum-Optimized Network Configuration', fontweight='bold')
    
    # Plot all nodes
    for i in range(quantum_optimizer.num_nodes):
        x, y = quantum_optimizer.coordinates[i]
        
        if i in quantum_solution['hub_configuration']:
            # Plot hub with quantum optimization indicator
            ax3.scatter(x, y, s=400, c='#FF6B6B', marker='s', 
                      edgecolors='black', linewidth=2, alpha=0.8)
            ax3.text(x+2, y+2, f'Q{i+1}', fontweight='bold', fontsize=8, color='red')
        else:
            # Plot spoke
            ax3.scatter(x, y, s=100, c='lightblue', 
                      edgecolors='black', linewidth=1, alpha=0.8)
            ax3.text(x+2, y+2, f'{i+1}', fontweight='bold', fontsize=8)
    
    # Plot hub connections
    for i, hub1 in enumerate(quantum_solution['hub_configuration']):
        for hub2 in quantum_solution['hub_configuration'][i+1:]:
            x1, y1 = quantum_optimizer.coordinates[hub1]
            x2, y2 = quantum_optimizer.coordinates[hub2]
            ax3.plot([x1, x2], [y1, y2], '-', color='red', 
                    linewidth=2, alpha=0.6)
    
    # Plot spoke assignments
    for node, hub in quantum_solution['spoke_assignments'].items():
        if node != hub:
            node_x, node_y = quantum_optimizer.coordinates[node]
            hub_x, hub_y = quantum_optimizer.coordinates[hub]
            ax3.plot([node_x, hub_x], [node_y, hub_y], '--', 
                    color='gray', alpha=0.4)
    
    ax3.set_xlabel('Longitude')
    ax3.set_ylabel('Latitude')
    ax3.grid(True, alpha=0.3)
    
    # Add quantum advantage annotation
    ax3.text(0.02, 0.98, 'Quantum Optimized Network', 
            transform=ax3.transAxes, va='top', fontweight='bold',
            bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.8))
    
    # 4. Solution Distribution
    ax4.set_title('Quantum Solution Distribution', fontweight='bold')
    
    # Show top solutions from quantum measurement
    top_solutions = sorted(quantum_result.solution_counts.items(), 
                          key=lambda x: x[1], reverse=True)[:10]
    
    if top_solutions:
        solution_strings = [f'Sol {i+1}' for i in range(len(top_solutions))]
        solution_counts = [count for _, count in top_solutions]
        
        bars = ax4.bar(solution_strings, solution_counts, color='#4ECDC4', alpha=0.8)
        ax4.set_ylabel('Measurement Count')
        ax4.set_xlabel('Solution')
        ax4.grid(True, alpha=0.3)
        
        # Highlight best solution
        ax4.bar(solution_strings[0], solution_counts[0], color='#FF6B6B', alpha=0.8)
        ax4.text(solution_strings[0], solution_counts[0] + max(solution_counts)*0.02, 
                'Best', ha='center', fontweight='bold', color='red')
    else:
        ax4.text(0.5, 0.5, 'No solution data available', 
                ha='center', va='center', transform=ax4.transAxes, fontsize=12)
    
    plt.tight_layout()
    plt.show()
    
    return fig

# Create quantum optimization visualization
fig = visualize_quantum_optimization(quantum_optimizer, quantum_solution, quantum_result)
```
