# 87. The Hub-and-Spoke Network Design Problem

## Tier 5: The Integrated Digital Twin (System-of-Systems Simulation)

### Key assumptions
- Real-time data streams from IoT sensors, weather stations, and tracking systems
- Physical-virtual synchronization maintains accurate network state representation
- Predictive analytics forecast demand fluctuations and disruption events
- Microservices architecture enables seamless subsystem integration
- What-if scenario analysis supports proactive decision making

### Approach (step-by-step)
1. **Data Integration**: Aggregate real-time data from multiple IoT sources
2. **State Synchronization**: Maintain virtual-physical system alignment
3. **Predictive Analytics**: Forecast demand, disruptions, and capacity changes
4. **Real-time Optimization**: Continuously update network configurations
5. **Scenario Simulation**: Test alternative routing and hub strategies
6. **Performance Monitoring**: Track KPIs and system health metrics

### What to look for in the results
- Real-time monitoring dashboard with live network status
- Predictive analytics accuracy and forecasting performance
- Disruption detection and automated response capabilities
- What-if scenario analysis results and recommendations
- System performance metrics and KPI tracking

### Concrete example (from the source)
We'll implement the MegaShip Global 15-port Pacific network example where the digital twin processes 2.3 million data points every 30 minutes, detects typhoon disruptions, and reduces delivery delays from 4.2 days to 1.8 days, saving $2.4 million in penalty costs.

```python
from typing import Tuple, List, Dict, Set
from itertools import product

# Import required libraries for Digital Twin implementation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import random
from datetime import datetime, timedelta
from typing import List, Tuple, Dict, Optional
import warnings
warnings.filterwarnings('ignore')

# Set professional styling for visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
try:
    # Define the Integrated Digital Twin framework for Hub-and-Spoke Networks
    class HubSpokeDigitalTwin:
        """Comprehensive Digital Twin for hub-and-spoke network management"""
        
        def __init__(self, num_ports=15, update_interval_minutes=30):
            self.num_ports = num_ports
            self.update_interval = update_interval_minutes
            self.current_time = datetime.now()
            
            # Network configuration
            self.port_coordinates = self._generate_port_coordinates()
            self.hub_ports = [2, 7, 12]  # Initial hub configuration
            self.port_capacities = np.random.randint(8000, 15000, num_ports)
            self.port_costs = np.random.randint(50000, 150000, num_ports)
            
            # Digital twin components
            self.iot_sensors = self._initialize_iot_sensors()
            self.weather_stations = self._initialize_weather_stations()
            self.vessel_tracking = self._initialize_vessel_tracking()
            self.demand_forecaster = DemandForecaster(num_ports)
            self.disruption_detector = DisruptionDetector(num_ports)
            
            # Real-time data storage
            self.current_network_state = {
                'port_utilization': np.zeros(num_ports),
                'vessel_positions': {},
                'weather_conditions': {},
                'demand_forecasts': np.zeros((num_ports, num_ports, 24)),  # 24-hour forecast
                'disruption_alerts': []
                'performance_kpis': {}
            }
            
            # Historical data for analysis
            self.historical_data = {
                'timestamps': [],
                'network_states': [],
                'optimization_decisions': [],
                'performance_metrics': []
            }
            
            # Optimization engine
            self.optimization_engine = RealTimeOptimizer(num_ports)
            
        def _generate_port_coordinates(self):
            """Generate realistic port coordinates for Pacific region"""
            # Simplified Pacific region coordinates
            coordinates = []
            
            # West Coast North America
            for i in range(5):
                coordinates.append([random.uniform(-125, -115), random.uniform(32, 49)])
            
            # Pacific Islands
            for i in range(5):
                coordinates.append([random.uniform(-180, -140), random.uniform(-20, 20)])
            
            # East Asia
            for i in range(5):
                coordinates.append([random.uniform(120, 145), random.uniform(20, 45)])
            
            return np.array(coordinates[:self.num_ports])
        
        def _initialize_iot_sensors(self):
            """Initialize IoT sensors for each port"""
            sensors = {}
            
            for port_id in range(self.num_ports):
                sensors[port_id] = {
                    'container_sensors': random.randint(20, 50),
                    'crane_sensors': random.randint(5, 15),
                    'gate_sensors': random.randint(10, 25),
                    'environmental_sensors': random.randint(5, 10),
                    'status': 'active',
                    'last_update': self.current_time
                }
            
            return sensors
        
        def _initialize_weather_stations(self):
            """Initialize weather monitoring stations"""
            stations = {}
            
            for port_id in range(self.num_ports):
                stations[port_id] = {
                    'wind_speed': random.uniform(0, 25),  # knots
                    'wave_height': random.uniform(0, 5),   # meters
                    'visibility': random.uniform(1, 10),   # km
                    'temperature': random.uniform(10, 35), # Celsius
                    'pressure': random.uniform(980, 1030), # millibar
                    'storm_probability': random.uniform(0, 0.3),
                    'last_update': self.current_time
                }
            
            return stations
        
        def _initialize_vessel_tracking(self):
            """Initialize vessel tracking system"""
            vessels = {}
            
            # Create active vessels
            num_vessels = random.randint(30, 50)
            
            for vessel_id in range(num_vessels):
                vessels[vessel_id] = {
                    'name': f'Vessel_{vessel_id+1}',
                    'type': random.choice(['Container', 'Bulk', 'Tanker']),
                    'capacity': random.randint(1000, 8000),  # TEU
                    'current_position': self._generate_random_position(),
                    'destination': random.randint(0, self.num_ports-1),
                    'eta': self.current_time + timedelta(hours=random.randint(6, 72)),
                    'status': random.choice(['in_transit', 'at_port', 'loading', 'unloading']),
                    'speed': random.uniform(10, 25),  # knots
                    'last_update': self.current_time
                }
            
            return vessels
        
        def _generate_random_position(self):
            """Generate random position in Pacific region"""
            lat = random.uniform(-50, 60)
            lon = random.uniform(-180, 180)
            return {'latitude': lat, 'longitude': lon}
        
        def collect_real_time_data(self):
            """Collect and process real-time data from all sources"""
            
            # Simulate data collection from IoT sensors
            for port_id in range(self.num_ports):
                # Update port utilization
                utilization = random.uniform(0.4, 0.95)
                self.current_network_state['port_utilization'][port_id] = utilization
                
                # Update weather conditions
                weather = self.weather_stations[port_id]
                weather['wind_speed'] += random.uniform(-2, 2)
                weather['wave_height'] += random.uniform(-0.5, 0.5)
                weather['storm_probability'] = max(0, min(1, 
                    weather['storm_probability'] + random.uniform(-0.05, 0.05)))
                weather['last_update'] = self.current_time
                
                self.current_network_state['weather_conditions'][port_id] = weather
            
            # Update vessel positions
            for vessel_id, vessel in self.vessel_tracking.items():
                if vessel['status'] == 'in_transit':
                    # Simulate vessel movement
                    vessel['current_position'] = self._update_vessel_position(vessel)
                    vessel['last_update'] = self.current_time
                
                self.current_network_state['vessel_positions'][vessel_id] = vessel
            
            # Update demand forecasts
            demand_forecast = self.demand_forecaster.generate_forecast()
            self.current_network_state['demand_forecasts'] = demand_forecast
            
            # Detect disruptions
            disruptions = self.disruption_detector.detect_disruptions(
                self.current_network_state['weather_conditions'],
                self.current_network_state['port_utilization']
            )
            self.current_network_state['disruption_alerts'] = disruptions
            
            # Calculate performance KPIs
            self.current_network_state['performance_kpis'] = self._calculate_kpis()
        
        def _update_vessel_position(self, vessel):
            """Update vessel position based on speed and course"""
            current_pos = vessel['current_position']
            
            # Simple movement simulation (would use actual navigation in production)
            lat_change = random.uniform(-0.5, 0.5) * vessel['speed'] / 20
            lon_change = random.uniform(-0.5, 0.5) * vessel['speed'] / 20
            
            new_position = {
                'latitude': current_pos['latitude'] + lat_change,
                'longitude': current_pos['longitude'] + lon_change
            }
            
            return new_position
        
        def _calculate_kpis(self):
            """Calculate current network performance KPIs"""
            
            kpis = {
                'average_utilization': np.mean(self.current_network_state['port_utilization']),
                'total_vessels': len(self.vessel_tracking),
                'vessels_at_port': sum(1 for v in self.vessel_tracking.values() 
                                   if v['status'] == 'at_port'),
                'active_disruptions': len(self.current_network_state['disruption_alerts']),
                'network_efficiency': self._calculate_network_efficiency(),
                'service_level': self._calculate_service_level(),
                'cost_per_container': self._calculate_cost_per_container(),
                'carbon_footprint': self._calculate_carbon_footprint()
            }
            
            return kpis
        
        def _calculate_network_efficiency(self):
            """Calculate overall network efficiency"""
            # Simplified efficiency calculation
            utilization_efficiency = np.mean(self.current_network_state['port_utilization'])
            
            # Weather impact on efficiency
            weather_impact = 0.0
            for weather in self.current_network_state['weather_conditions'].values():
                if weather['storm_probability'] > 0.2:
                    weather_impact += weather['storm_probability'] * 0.1
            
            weather_impact /= self.num_ports
            
            efficiency = utilization_efficiency * (1 - weather_impact)
            return max(0, min(1, efficiency))
        
        def _calculate_service_level(self):
            """Calculate service level percentage"""
            # Simplified service level based on utilization and disruptions
            base_service = 0.95
            utilization_penalty = max(0, np.mean(self.current_network_state['port_utilization']) - 0.8) * 0.5
            disruption_penalty = min(0.2, len(self.current_network_state['disruption_alerts']) * 0.05)
            
            service_level = base_service - utilization_penalty - disruption_penalty
            return max(0, min(1, service_level))
        
        def _calculate_cost_per_container(self):
            """Calculate average cost per container"""
            base_cost = 150  # Base cost per container
            utilization_factor = 1 + np.mean(self.current_network_state['port_utilization']) * 0.3
            weather_surcharge = sum(w['storm_probability'] for w in 
                                  self.current_network_state['weather_conditions'].values()) * 10
            
            total_cost = base_cost * utilization_factor + weather_surcharge / self.num_ports
            return total_cost
        
        def _calculate_carbon_footprint(self):
            """Calculate carbon footprint in tons CO2"""
            # Simplified carbon calculation
            vessel_emissions = len(self.vessel_tracking) * 50  # tons per vessel per update
            port_emissions = np.sum(self.current_network_state['port_utilization']) * 100  # tons per port
            
            total_emissions = vessel_emissions + port_emissions
            return total_emissions
        
        def run_real_time_optimization(self):
            """Run real-time optimization based on current state"""
            
            # Get optimization recommendations
            recommendations = self.optimization_engine.optimize(
                self.current_network_state,
                self.hub_ports,
                self.current_network_state['disruption_alerts']
            )
            
            # Apply recommendations if they improve performance
            if recommendations['expected_improvement'] > 0.05:  # 5% improvement threshold
                self.hub_ports = recommendations['new_hub_configuration']
                
                # Record decision
                decision = {
                    'timestamp': self.current_time,
                    'type': 'hub_reconfiguration',
                    'old_hubs': recommendations['old_hub_configuration'],
                    'new_hubs': recommendations['new_hub_configuration'],
                    'expected_improvement': recommendations['expected_improvement'],
                    'reason': recommendations['reason']
                }
                
                self.historical_data['optimization_decisions'].append(decision)
            
            return recommendations
        
        def simulate_what_if_scenarios(self):
            """Simulate what-if scenarios for strategic planning"""
            
            scenarios = {
                'typhoon_disruption': self._simulate_typhoon_scenario(),
                'demand_surge': self._simulate_demand_surge(),
                'new_hub_addition': self._simulate_new_hub(),
                'capacity_expansion': self._simulate_capacity_expansion()
            }
            
            return scenarios
        
        def _simulate_typhoon_scenario(self):
            """Simulate typhoon disruption scenario"""
            
            # Select a hub to be affected by typhoon
            affected_hub = random.choice(self.hub_ports)
            
            # Simulate capacity degradation
            capacity_reduction = random.uniform(0.3, 0.7)  # 30-70% capacity loss
            duration_hours = random.randint(24, 72)
            
            # Calculate impact
            original_capacity = self.port_capacities[affected_hub]
            reduced_capacity = original_capacity * (1 - capacity_reduction)
            
            # Find alternative routing
            alternative_hubs = [h for h in self.hub_ports if h != affected_hub]
            if alternative_hubs:
                alternative_hub = min(alternative_hubs, 
                                   key=lambda h: self._calculate_distance(affected_hub, h))
            else:
                alternative_hub = None
            
            return {
                'scenario_type': 'typhoon_disruption',
                'affected_hub': affected_hub,
                'capacity_reduction': capacity_reduction,
                'duration_hours': duration_hours,
                'original_capacity': original_capacity,
                'reduced_capacity': reduced_capacity,
                'alternative_hub': alternative_hub,
                'estimated_cost_impact': capacity_reduction * original_capacity * 0.1,
                'recommended_actions': self._generate_typhoon_response(affected_hub, alternative_hub)
            }
        
        def _simulate_demand_surge(self):
            """Simulate demand surge scenario"""
            
            # Random demand increase
            demand_increase = random.uniform(1.2, 1.5)  # 20-50% increase
            affected_regions = random.sample(range(self.num_ports), random.randint(3, 7))
            
            # Calculate impact on hub utilization
            current_utilization = np.mean(self.current_network_state['port_utilization'])
            projected_utilization = min(1.0, current_utilization * demand_increase)
            
            return {
                'scenario_type': 'demand_surge',
                'demand_increase': demand_increase,
                'affected_regions': affected_regions,
                'current_utilization': current_utilization,
                'projected_utilization': projected_utilization,
                'capacity_risk': projected_utilization > 0.9,
                'recommended_actions': self._generate_demand_surge_response(demand_increase)
            }
        
        def _simulate_new_hub(self):
            """Simulate adding a new hub to the network"""
            
            # Find best candidate for new hub
            non_hubs = [i for i in range(self.num_ports) if i not in self.hub_ports]
            
            if not non_hubs:
                return {'scenario_type': 'new_hub_addition', 'status': 'no_candidates'}
            
            # Evaluate candidates based on centrality and capacity
            best_candidate = max(non_hubs, key=lambda h: self._evaluate_hub_candidate(h))
            
            # Calculate expected benefits
            current_efficiency = self.current_network_state['performance_kpis']['network_efficiency']
            projected_efficiency = min(1.0, current_efficiency * 1.1)  # 10% improvement
            
            return {
                'scenario_type': 'new_hub_addition',
                'candidate_hub': best_candidate,
                'candidate_score': self._evaluate_hub_candidate(best_candidate),
                'current_efficiency': current_efficiency,
                'projected_efficiency': projected_efficiency,
                'implementation_cost': self.port_costs[best_candidate],
                'payback_period_months': self._calculate_payback_period(best_candidate),
                'recommended': projected_efficiency - current_efficiency > 0.05
            }
        
        def _simulate_capacity_expansion(self):
            """Simulate capacity expansion at existing hubs"""
            
            # Select hub with highest utilization
            utilization_by_hub = {}
            for hub in self.hub_ports:
                utilization_by_hub[hub] = self.current_network_state['port_utilization'][hub]
            
            if not utilization_by_hub:
                return {'scenario_type': 'capacity_expansion', 'status': 'no_hubs'}
            
            highest_utilization_hub = max(utilization_by_hub, key=utilization_by_hub.get)
            current_capacity = self.port_capacities[highest_utilization_hub]
            
            # Expansion options
            expansion_options = [0.2, 0.3, 0.5]  # 20%, 30%, 50% expansion
            
            results = []
            for expansion in expansion_options:
                new_capacity = current_capacity * (1 + expansion)
                expansion_cost = new_capacity * 50  # $50 per unit capacity
                
                # Calculate benefits
                current_util = utilization_by_hub[highest_utilization_hub]
                new_util = current_util / (1 + expansion)
                efficiency_gain = (current_util - new_util) * 0.5  # Simplified efficiency calculation
                
                results.append({
                    'expansion_percentage': expansion,
                    'new_capacity': new_capacity,
                    'expansion_cost': expansion_cost,
                    'current_utilization': current_util,
                    'projected_utilization': new_util,
                    'efficiency_gain': efficiency_gain,
                    'roi': (efficiency_gain * 10000) / expansion_cost if expansion_cost > 0 else 0
                })
            
            return {
                'scenario_type': 'capacity_expansion',
                'target_hub': highest_utilization_hub,
                'current_capacity': current_capacity,
                'current_utilization': utilization_by_hub[highest_utilization_hub],
                'expansion_options': results,
                'recommended_option': max(results, key=lambda x: x['roi']) if results else None
            }
        
        def _calculate_distance(self, port1, port2):
            """Calculate distance between two ports"""
            coord1 = self.port_coordinates[port1]
            coord2 = self.port_coordinates[port2]
            return np.linalg.norm(coord1 - coord2)
        
        def _evaluate_hub_candidate(self, port_id):
            """Evaluate a port as a potential hub"""
            
            # Factors: centrality, capacity, cost, utilization
            centrality_score = self._calculate_centrality(port_id)
            capacity_score = self.port_capacities[port_id] / np.max(self.port_capacities)
            cost_score = 1 - (self.port_costs[port_id] / np.max(self.port_costs))
            utilization_score = 1 - self.current_network_state['port_utilization'][port_id]
            
            total_score = (centrality_score * 0.3 + capacity_score * 0.3 + 
                           cost_score * 0.2 + utilization_score * 0.2)
            
            return total_score
        
        def _calculate_centrality(self, port_id):
            """Calculate network centrality for a port"""
            distances = [self._calculate_distance(port_id, other) 
                       for other in range(self.num_ports) if other != port_id]
            
            if distances:
                avg_distance = np.mean(distances)
                return 1 / (1 + avg_distance)  # Lower distance = higher centrality
            return 0.5
        
        def _calculate_payback_period(self, hub_id):
            """Calculate payback period for new hub in months"""
            # Simplified payback calculation
            investment = self.port_costs[hub_id]
            monthly_benefit = investment * 0.05  # 5% monthly return
            
            return investment / monthly_benefit if monthly_benefit > 0 else float('inf')
        
        def _generate_typhoon_response(self, affected_hub, alternative_hub):
            """Generate response recommendations for typhoon scenario"""
            
            actions = [
                f"Activate emergency response protocol for Hub {affected_hub+1}",
                f"Reroute {random.randint(20, 40)}% of traffic to alternative hubs"
            ]
            
            if alternative_hub is not None:
                actions.append(f"Increase Hub {alternative_hub+1} capacity utilization")
            
            actions.extend([
                "Notify affected customers of potential delays",
                "Activate contingency vessel contracts",
                "Update delivery time estimates"
            ])
            
            return actions
        
        def _generate_demand_surge_response(self, demand_increase):
            """Generate response recommendations for demand surge"""
            
            actions = [
                f"Activate surge capacity protocols ({demand_increase-1:.0%} increase)",
                "Extend operating hours at high-utilization ports",
                "Activate standby vessel fleet"
            ]
            
            if demand_increase > 1.3:
                actions.append("Consider temporary hub activation")
            
            actions.extend([
                "Implement priority routing for critical cargo",
                "Coordinate with alliance partners for capacity sharing"
            ])
            
            return actions
        
        def update_historical_data(self):
            """Update historical data with current state"""
            
            self.historical_data['timestamps'].append(self.current_time)
            self.historical_data['network_states'].append(self.current_network_state.copy())
            self.historical_data['performance_metrics'].append(self.current_network_state['performance_kpis'].copy())
            
            # Keep only last 1000 records to manage memory
            if len(self.historical_data['timestamps']) > 1000:
                for key in self.historical_data:
                    self.historical_data[key] = self.historical_data[key][-1000:]
    
    # Supporting classes for the Digital Twin
    class DemandForecaster:
        """Demand forecasting component"""
        
        def __init__(self, num_ports):
            self.num_ports = num_ports
            self.base_demand = np.random.randint(100, 500, (num_ports, num_ports))
            np.fill_diagonal(self.base_demand, 0)
        
        def generate_forecast(self):
            """Generate 24-hour demand forecast"""
            forecast = np.zeros((self.num_ports, self.num_ports, 24))
            
            for hour in range(24):
                # Add temporal variation and noise
                hourly_factor = 1 + 0.3 * np.sin(2 * np.pi * hour / 24)  # Daily pattern
                noise = np.random.normal(1, 0.1, (self.num_ports, self.num_ports))
                
                forecast[:, :, hour] = self.base_demand * hourly_factor * noise
                np.fill_diagonal(forecast[:, :, hour], 0)
            
            return forecast
    
    class DisruptionDetector:
        """Disruption detection component"""
        
        def __init__(self, num_ports):
            self.num_ports = num_ports
        
        def detect_disruptions(self, weather_conditions, port_utilization):
            """Detect potential disruptions"""
            
            disruptions = []
            
            for port_id in range(self.num_ports):
                weather = weather_conditions[port_id]
                utilization = port_utilization[port_id]
                
                # Weather-related disruptions
                if weather['storm_probability'] > 0.3:
                    disruptions.append({
                        'type': 'weather',
                        'port': port_id,
                        'severity': 'high' if weather['storm_probability'] > 0.6 else 'medium',
                        'description': f"Storm probability: {weather['storm_probability']:.2f}",
                        'timestamp': datetime.now()
                    })
                
                # Capacity-related disruptions
                if utilization > 0.9:
                    disruptions.append({
                        'type': 'capacity',
                        'port': port_id,
                        'severity': 'critical' if utilization > 0.95 else 'high',
                        'description': f"High utilization: {utilization:.2f}",
                        'timestamp': datetime.now()
                    })
            
            return disruptions
    
    class RealTimeOptimizer:
        """Real-time optimization engine"""
        
        def __init__(self, num_ports):
            self.num_ports = num_ports
        
        def optimize(self, network_state, current_hubs, disruptions):
            """Generate optimization recommendations"""
            
            # Check if reconfiguration is needed
            reconfigure_needed = False
            reason = ""
            
            # Check for disruptions affecting current hubs
            for disruption in disruptions:
                if disruption['port'] in current_hubs and disruption['severity'] in ['high', 'critical']:
                    reconfigure_needed = True
                    reason = f"Disruption at Hub {disruption['port']+1}: {disruption['description']}"
                    break
            
            # Check for utilization issues
            avg_utilization = np.mean(network_state['port_utilization'])
            if avg_utilization > 0.85:
                reconfigure_needed = True
                reason = f"High network utilization: {avg_utilization:.2f}"
            
            if reconfigure_needed:
                # Generate new hub configuration
                new_hubs = self._generate_new_hub_configuration(network_state, disruptions)
                
                # Calculate expected improvement
                expected_improvement = random.uniform(0.05, 0.15)  # 5-15% improvement
                
                return {
                    'action': 'reconfigure',
                    'old_hub_configuration': current_hubs.copy(),
                    'new_hub_configuration': new_hubs,
                    'expected_improvement': expected_improvement,
                    'reason': reason,
                    'implementation_time': '30 minutes'
                }
            
            return {
                'action': 'maintain',
                'old_hub_configuration': current_hubs.copy(),
                'new_hub_configuration': current_hubs.copy(),
                'expected_improvement': 0.0,
                'reason': 'Current configuration optimal',
                'implementation_time': 'N/A'
            }
        
        def _generate_new_hub_configuration(self, network_state, disruptions):
            """Generate new hub configuration"""
            
            # Get available ports (excluding disrupted ones)
            disrupted_ports = {d['port'] for d in disruptions if d['severity'] in ['high', 'critical']}
            available_ports = [i for i in range(self.num_ports) if i not in disrupted_ports]
            
            if len(available_ports) >= 3:
                # Select top 3 ports based on utilization and capacity
                port_scores = []
                for port in available_ports:
                    utilization = network_state['port_utilization'][port]
                    score = (1 - utilization) * 0.7 + random.uniform(0, 0.3)  # Prefer lower utilization
                    port_scores.append((port, score))
                
                port_scores.sort(key=lambda x: x[1], reverse=True)
                new_hubs = [port for port, score in port_scores[:3]]
            else:
                new_hubs = available_ports[:min(3, len(available_ports))]
            
            return new_hubs
    
    # Initialize the Digital Twin system
    print("Initializing Hub-and-Spoke Digital Twin System...")
    digital_twin = HubSpokeDigitalTwin(num_ports=15, update_interval_minutes=30)
    
    print(f"Digital Twin Configuration:")
    print(f"- Number of ports: {digital_twin.num_ports}")
    print(f"- Initial hub ports: {[h+1 for h in digital_twin.hub_ports]}")
    print(f"- Update interval: {digital_twin.update_interval} minutes")
    print(f"- IoT sensors: {sum(len(sensors) for sensors in digital_twin.iot_sensors.values())}")
    print(f"- Weather stations: {len(digital_twin.weather_stations)}")
    print(f"- Tracked vessels: {len(digital_twin.vessel_tracking)}")
except Exception as e:
    print(f'Error in cell: {e}')
```
