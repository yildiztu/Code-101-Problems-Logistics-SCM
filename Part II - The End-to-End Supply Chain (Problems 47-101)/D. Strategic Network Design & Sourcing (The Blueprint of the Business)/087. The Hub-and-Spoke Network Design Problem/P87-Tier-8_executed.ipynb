# 87. The Hub-and-Spoke Network Design Problem

## Tier 8: The Value-Aligned & Ethical Framework (Constitutional AI)

### Key assumptions
- Network optimization must align with ethical principles and stakeholder values
- Multiple stakeholders have competing interests that require balanced consideration
- Environmental sustainability, social responsibility, and economic viability are equally important
- Ethical constraints must be embedded into the optimization process, not added as afterthoughts
- Transparency and accountability are essential for ethical decision-making

### Approach (step-by-step)
1. **Stakeholder Identification**: Identify all relevant stakeholder groups and their interests
2. **Ethical Framework Definition**: Establish constitutional principles and ethical constraints
3. **Multi-Objective Optimization**: Balance competing objectives with ethical constraints
4. **Value Alignment**: Ensure solutions align with organizational and societal values
5. **Impact Assessment**: Evaluate social, environmental, and economic impacts
6. **Ethical Compliance**: Verify all decisions meet ethical and regulatory requirements

### What to look for in the results
- Multi-stakeholder utility optimization with balanced outcomes
- Ethical constraint satisfaction and compliance metrics
- Environmental impact reduction and sustainability improvements
- Social responsibility indicators and community impact assessments
- Trade-off analysis between competing ethical objectives

### Concrete example (from the source)
We'll implement the Global Logistics Alliance 20-node network example where the ethical framework balances shareholder returns, employee welfare, environmental sustainability, and community impact, achieving 27% carbon reduction while maintaining 95% service levels and ensuring equitable treatment of all stakeholders.

```python
try:
    try:
        # Define the Value-Aligned & Ethical Framework
        @dataclass
        class StakeholderGroup:
            """Represents a stakeholder group with their interests and weights"""
            name: str
            interests: List[str]
            utility_weights: Dict[str, float]  # weights for different objectives
            constraints: List[str]
            priority: float  # relative priority (0-1)
            
        @dataclass
        class EthicalPrinciple:
            """Represents an ethical principle with its constraints and metrics"""
            name: str
            description: str
            constraints: List[str]
            metrics: List[str]
            minimum_threshold: float
            penalty_weight: float
            
        @dataclass
        class EthicalConstraint:
            """Represents an ethical constraint that must be satisfied"""
            name: str
            type: str  # 'hard', 'soft', 'preference'
            description: str
            evaluation_function: str
            threshold: float
            penalty_factor: float
            
        @dataclass
        class ValueAlignedSolution:
            """Represents a value-aligned network solution"""
            hub_configuration: List[int]
            spoke_assignments: Dict[int, int]
            stakeholder_utilities: Dict[str, float]
            ethical_compliance: Dict[str, float]
            overall_score: float
            trade_offs: Dict[str, float]
            explanation: str
        
        class ValueAlignedEthicalFramework:
            """Value-Aligned & Ethical Framework for hub-and-spoke network design"""
            
            def __init__(self, num_nodes=20):
                self.num_nodes = num_nodes
                
                # Initialize stakeholder groups
                self.stakeholders = self._initialize_stakeholders()
                
                # Initialize ethical principles
                self.ethical_principles = self._initialize_ethical_principles()
                
                # Initialize ethical constraints
                self.ethical_constraints = self._initialize_ethical_constraints()
                
                # Initialize network data
                self._initialize_network_data()
                
                # Optimization engine
                self.optimization_engine = EthicalOptimizationEngine()
                
                # Impact assessment system
                self.impact_assessor = ImpactAssessmentSystem()
                
                # Compliance monitor
                self.compliance_monitor = EthicalComplianceMonitor()
                
                # Solution history
                self.solution_history = []
                self.compliance_history = []
            
            def _initialize_stakeholders(self):
                """Initialize stakeholder groups with their interests and weights"""
                
                stakeholders = {
                    'shareholders': StakeholderGroup(
                        name='Shareholders',
                        interests=['profitability', 'market_share', 'dividend_growth', 'stock_performance'],
                        utility_weights={
                            'cost_efficiency': 0.4,
                            'revenue_growth': 0.3,
                            'market_expansion': 0.2,
                            'risk_management': 0.1
                        },
                        constraints=['minimum_roi', 'debt_ratio_limit', 'compliance_requirements'],
                        priority=0.25
                    ),
                    'employees': StakeholderGroup(
                        name='Employees',
                        interests=['job_security', 'fair_wages', 'working_conditions', 'career_development'],
                        utility_weights={
                            'employment_stability': 0.35,
                            'wage_fairness': 0.25,
                            'work_life_balance': 0.2,
                            'safety_standards': 0.15,
                            'training_opportunities': 0.05
                        },
                        constraints=['minimum_wage', 'safety_standards', 'working_hours'],
                        priority=0.20
                    ),
                    'customers': StakeholderGroup(
                        name='Customers',
                        interests=['service_quality', 'price_competitiveness', 'reliability', 'sustainability'],
                        utility_weights={
                            'service_quality': 0.3,
                            'price_competitiveness': 0.25,
                            'delivery_reliability': 0.25,
                            'product_quality': 0.15,
                            'environmental_impact': 0.05
                        },
                        constraints=['service_level_agreement', 'quality_standards', 'fair_pricing'],
                        priority=0.25
                    ),
                    'environment': StakeholderGroup(
                        name='Environment',
                        interests=['carbon_reduction', 'waste_minimization', 'biodiversity', 'renewable_energy'],
                        utility_weights={
                            'carbon_emissions': 0.35,
                            'waste_generation': 0.25,
                            'energy_efficiency': 0.2,
                            'resource_conservation': 0.15,
                            'ecosystem_protection': 0.05
                        },
                        constraints=['emission_limits', 'waste_disposal_regulations', 'renewable_energy_targets'],
                        priority=0.15
                    ),
                    'community': StakeholderGroup(
                        name='Community',
                        interests=['local_employment', 'economic_development', 'social_impact', 'infrastructure'],
                        utility_weights={
                            'local_jobs': 0.3,
                            'economic_contribution': 0.25,
                            'social_welfare': 0.2,
                            'infrastructure_development': 0.15,
                            'cultural_preservation': 0.1
                        },
                        constraints=['local_hiring_requirements', 'community_investment', 'noise_pollution_limits'],
                        priority=0.15
                    )
                }
                
                return stakeholders
            
            def _initialize_ethical_principles(self):
                """Initialize ethical principles with constraints and metrics"""
                
                principles = {
                    'sustainability': EthicalPrinciple(
                        name='Sustainability',
                        description='Minimize environmental impact while maintaining economic viability',
                        constraints=['carbon_neutrality', 'waste_reduction', 'renewable_energy'],
                        metrics=['carbon_footprint', 'waste_generated', 'energy_consumption'],
                        minimum_threshold=0.8,
                        penalty_weight=0.3
                    ),
                    'equity': EthicalPrinciple(
                        name='Equity',
                        description='Ensure fair treatment of all stakeholders and communities',
                        constraints=['fair_wages', 'equal_opportunity', 'community_benefits'],
                        metrics=['wage_distribution', 'employment_diversity', 'community_impact'],
                        minimum_threshold=0.7,
                        penalty_weight=0.25
                    ),
                    'transparency': EthicalPrinciple(
                        name='Transparency',
                        description='Maintain openness and accountability in decision-making',
                        constraints=['disclosure_requirements', 'stakeholder_communication', 'audit_trails'],
                        metrics=['information_disclosure', 'decision_transparency', 'reporting_quality'],
                        minimum_threshold=0.9,
                        penalty_weight=0.2
                    ),
                    'responsibility': EthicalPrinciple(
                        name='Responsibility',
                        description='Take responsibility for social and environmental impacts',
                        constraints=['liability_coverage', 'remediation_funds', 'safety_standards'],
                        metrics=['safety_record', 'incident_response', 'remediation_effectiveness'],
                        minimum_threshold=0.85,
                        penalty_weight=0.25
                    )
                }
                
                return principles
            
            def _initialize_ethical_constraints(self):
                """Initialize specific ethical constraints"""
                
                constraints = [
                    EthicalConstraint(
                        name='maximum_carbon_emissions',
                        type='hard',
                        description='Carbon emissions must not exceed regulatory limits',
                        evaluation_function='calculate_total_emissions',
                        threshold=10000,  # tons per year
                        penalty_factor=10.0
                    ),
                    EthicalConstraint(
                        name='minimum_service_level',
                        type='hard',
                        description='Service level must meet customer requirements',
                        evaluation_function='calculate_service_level',
                        threshold=0.95,
                        penalty_factor=8.0
                    ),
                    EthicalConstraint(
                        name='fair_wage_distribution',
                        type='soft',
                        description='Wage distribution must be equitable across regions',
                        evaluation_function='calculate_wage_equity',
                        threshold=0.8,
                        penalty_factor=3.0
                    ),
                    EthicalConstraint(
                        name='local_employment_minimum',
                        type='soft',
                        description='Minimum percentage of local employment required',
                        evaluation_function='calculate_local_employment',
                        threshold=0.6,
                        penalty_factor=4.0
                    ),
                    EthicalConstraint(
                        name='community_investment_minimum',
                        type='preference',
                        description='Minimum community investment percentage',
                        evaluation_function='calculate_community_investment',
                        threshold=0.02,
                        penalty_factor=2.0
                    )
                ]
                
                return constraints
            
            def _initialize_network_data(self):
                """Initialize network data for the 20-node global network"""
                
                # Generate global coordinates (simplified world map)
                self.coordinates = self._generate_global_coordinates()
                
                # Generate demand matrix
                self.demand_matrix = self._generate_global_demand_matrix()
                
                # Generate distance matrix
                self.distance_matrix = self._calculate_global_distance_matrix()
                
                # Generate cost and capacity data
                self.hub_costs = self._generate_global_hub_costs()
                self.capacity_limits = self._generate_global_capacity_limits()
                
                # Generate environmental and social data
                self.emission_factors = self._generate_emission_factors()
                self.social_impact_factors = self._generate_social_impact_factors()
                
                # Generate stakeholder-specific data
                self.wage_levels = self._generate_wage_levels()
                self.community_data = self._generate_community_data()
            
            def _generate_global_coordinates(self):
                """Generate coordinates for global network"""
                
                coordinates = []
                
                # North America
                coordinates.extend([
                    [-74.0, 40.7],   # New York
                    [-118.2, 34.1], # Los Angeles
                    [-87.6, 41.9],  # Chicago
                    [-122.3, 47.6], # Seattle
                ])
                
                # Europe
                coordinates.extend([
                    [2.3, 48.9],    # Rotterdam
                    [9.9, 53.6],    # Hamburg
                    [-0.1, 51.5],   # London
                    [12.5, 55.7],   # Copenhagen
                ])
                
                # Asia
                coordinates.extend([
                    [121.5, 31.2],  # Shanghai
                    [139.7, 35.7],  # Tokyo
                    [103.8, 1.3],   # Singapore
                    [114.2, 22.3],  # Hong Kong
                ])
                
                # South America
                coordinates.extend([
                    [-46.6, -23.5], # São Paulo
                    [-70.7, -33.5], # Santiago
                ])
                
                # Africa
                coordinates.extend([
                    [18.4, -33.9],  # Cape Town
                    [31.2, 30.1],   # Cairo
                ])
                
                # Oceania
                coordinates.extend([
                    [151.2, -33.9], # Sydney
                ])
                
                return np.array(coordinates[:self.num_nodes])
            
            def _generate_global_demand_matrix(self):
                """Generate realistic global demand matrix"""
                
                demand_matrix = np.zeros((self.num_nodes, self.num_nodes))
                
                # Regional demand patterns
                regional_multipliers = {
                    'north_america': 1.2,
                    'europe': 1.1,
                    'asia': 1.3,
                    'south_america': 0.8,
                    'africa': 0.6,
                    'oceania': 0.7
                }
                
                for i in range(self.num_nodes):
                    for j in range(self.num_nodes):
                        if i != j:
                            # Base demand
                            base_demand = random.randint(100, 800)
                            
                            # Apply regional multipliers
                            region_i = self._get_node_region(i)
                            region_j = self._get_node_region(j)
                            
                            multiplier = regional_multipliers.get(region_i, 1.0) * regional_multipliers.get(region_j, 1.0)
                            
                            # Distance penalty
                            distance = self.distance_matrix[i][j]
                            distance_factor = 1.0 / (1 + distance / 5000)
                            
                            demand_matrix[i][j] = int(base_demand * multiplier * distance_factor)
                
                return demand_matrix
            
            def _get_node_region(self, node_id):
                """Get region for a node based on its coordinates"""
                
                lon, lat = self.coordinates[node_id]
                
                if -125 <= lon <= -60 and 25 <= lat <= 50:
                    return 'north_america'
                elif -10 <= lon <= 40 and 35 <= lat <= 70:
                    return 'europe'
                elif 60 <= lon <= 150 and -10 <= lat <= 50:
                    return 'asia'
                elif -80 <= lon <= -35 and -55 <= lat <= 10:
                    return 'south_america'
                elif -20 <= lon <= 50 and -35 <= lat <= 40:
                    return 'africa'
                elif 110 <= lon <= 160 and -45 <= lat <= -10:
                    return 'oceania'
                else:
                    return 'other'
            
            def _calculate_global_distance_matrix(self):
                """Calculate global distance matrix using great circle distance"""
                
                distance_matrix = np.zeros((self.num_nodes, self.num_nodes))
                
                for i in range(self.num_nodes):
                    for j in range(self.num_nodes):
                        if i != j:
                            # Great circle distance calculation
                            lat1, lon1 = np.radians(self.coordinates[i][1]), np.radians(self.coordinates[i][0])
                            lat2, lon2 = np.radians(self.coordinates[j][1]), np.radians(self.coordinates[j][0])
                            
                            dlat = lat2 - lat1
                            dlon = lon2 - lon1
                            
                            a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
                            c = 2 * np.arcsin(np.sqrt(a))
                            
                            distance_matrix[i][j] = 6371 * c  # Earth radius in km
                
                return distance_matrix
            
            def _generate_global_hub_costs(self):
                """Generate hub costs for global network"""
                
                # Base costs with regional variations
                base_costs = np.random.uniform(150000, 500000, self.num_nodes)
                
                # Regional cost adjustments
                for i in range(self.num_nodes):
                    region = self._get_node_region(i)
                    
                    if region == 'north_america' or region == 'europe':
                        base_costs[i] *= 1.3  # Higher costs in developed regions
                    elif region == 'asia':
                        base_costs[i] *= 1.1  # Moderate costs
                    else:
                        base_costs[i] *= 0.7  # Lower costs in developing regions
                
                # Major hub premium
                major_hubs = [0, 1, 4, 8, 9]  # New York, LA, Rotterdam, Shanghai, Tokyo
                for hub in major_hubs:
                    if hub < self.num_nodes:
                        base_costs[hub] *= 1.5
                
                return base_costs.astype(int)
            
            def _generate_global_capacity_limits(self):
                """Generate capacity limits for global network"""
                
                base_capacities = np.random.uniform(15000, 50000, self.num_nodes)
                
                # Regional capacity adjustments
                for i in range(self.num_nodes):
                    region = self._get_node_region(i)
                    
                    if region == 'north_america' or region == 'europe':
                        base_capacities[i] *= 1.4  # Higher capacity in developed regions
                    elif region == 'asia':
                        base_capacities[i] *= 1.2  # Good capacity
                    else:
                        base_capacities[i] *= 0.8  # Lower capacity in developing regions
                
                return base_capacities.astype(int)
            
            def _generate_emission_factors(self):
                """Generate carbon emission factors for each route"""
                
                emission_factors = np.zeros((self.num_nodes, self.num_nodes))
                
                for i in range(self.num_nodes):
                    for j in range(self.num_nodes):
                        if i != j:
                            # Base emission factor (kg CO2 per container)
                            base_factor = 150  # Average for global routes
                            
                            # Distance-based adjustment
                            distance = self.distance_matrix[i][j]
                            distance_adjustment = distance / 1000  # Normalize to 1000km
                            
                            # Mode-based adjustment (sea vs air vs land)
                            if distance > 3000:  # Long routes likely sea
                                mode_adjustment = 0.6
                            elif distance < 500:  # Short routes likely land
                                mode_adjustment = 1.4
                            else:
                                mode_adjustment = 1.0
                            
                            emission_factors[i][j] = base_factor * distance_adjustment * mode_adjustment
                
                return emission_factors
            
            def _generate_social_impact_factors(self):
                """Generate social impact factors for each node"""
                
                social_factors = {}
                
                for i in range(self.num_nodes):
                    region = self._get_node_region(i)
                    
                    # Employment impact factor
                    if region in ['north_america', 'europe']:
                        employment_factor = 0.8  # Lower employment impact per container
                    elif region == 'asia':
                        employment_factor = 1.2  # Higher employment impact
                    else:
                        employment_factor = 1.5  # Highest employment impact
                    
                    # Community development factor
                    if region in ['north_america', 'europe']:
                        community_factor = 0.6  # Lower community need
                    else:
                        community_factor = 1.4  # Higher community need
                    
                    social_factors[i] = {
                        'employment_impact': employment_factor,
                        'community_development': community_factor,
                        'skill_development': 1.0,
                        'infrastructure_need': 1.0
                    }
                
                return social_factors
            
            def _generate_wage_levels(self):
                """Generate wage levels for each node"""
                
                wage_levels = np.zeros(self.num_nodes)
                
                for i in range(self.num_nodes):
                    region = self._get_node_region(i)
                    
                    if region == 'north_america':
                        wage_levels[i] = random.uniform(15, 25)  # $15-25/hour
                    elif region == 'europe':
                        wage_levels[i] = random.uniform(12, 22)  # $12-22/hour
                    elif region == 'asia':
                        wage_levels[i] = random.uniform(3, 8)    # $3-8/hour
                    elif region == 'south_america':
                        wage_levels[i] = random.uniform(4, 10)   # $4-10/hour
                    elif region == 'africa':
                        wage_levels[i] = random.uniform(2, 6)    # $2-6/hour
                    else:
                        wage_levels[i] = random.uniform(5, 15)   # $5-15/hour
                
                return wage_levels
            
            def _generate_community_data(self):
                """Generate community data for each node"""
                
                community_data = {}
                
                for i in range(self.num_nodes):
                    region = self._get_node_region(i)
                    
                    # Community need index
                    if region in ['north_america', 'europe']:
                        need_index = random.uniform(0.2, 0.4)
                    elif region == 'asia':
                        need_index = random.uniform(0.4, 0.6)
                    else:
                        need_index = random.uniform(0.6, 0.9)
                    
                    community_data[i] = {
                        'need_index': need_index,
                        'population_density': random.uniform(100, 10000),
                        'education_level': random.uniform(0.3, 0.9),
                        'infrastructure_quality': random.uniform(0.3, 0.9)
                    }
                
                return community_data
            
            def optimize_value_aligned_network(self):
                """Optimize network with value-aligned ethical constraints"""
                
                print("=== VALUE-ALIGNED ETHICAL OPTIMIZATION ===")
                
                # Step 1: Generate initial candidate solutions
                print("1. Generating candidate solutions...")
                candidate_solutions = self._generate_candidate_solutions()
                print(f"   Generated {len(candidate_solutions)} candidate solutions")
                
                # Step 2: Evaluate stakeholder utilities
                print("\n2. Evaluating stakeholder utilities...")
                evaluated_solutions = []
                
                for candidate in candidate_solutions:
                    utilities = self._calculate_stakeholder_utilities(candidate)
                    compliance = self._evaluate_ethical_compliance(candidate)
                    
                    evaluated_solutions.append({
                        'solution': candidate,
                        'utilities': utilities,
                        'compliance': compliance,
                        'overall_score': self._calculate_overall_score(utilities, compliance)
                        'trade_offs': self._analyze_trade_offs(utilities)
                    })
                
                # Step 3: Select best value-aligned solution
                print("\n3. Selecting best value-aligned solution...")
                best_solution = max(evaluated_solutions, key=lambda x: x['overall_score'])
                
                print(f"   Best solution score: {best_solution['overall_score']:.3f}")
                print(f"   Ethical compliance: {best_solution['compliance']['overall_compliance']:.1%}")
                
                # Step 4: Generate explanation
                explanation = self._generate_solution_explanation(best_solution)
                
                # Step 5: Create final solution
                final_solution = ValueAlignedSolution(
                    hub_configuration=best_solution['solution']['hub_configuration'],
                    spoke_assignments=best_solution['solution']['spoke_assignments'],
                    stakeholder_utilities=best_solution['utilities'],
                    ethical_compliance=best_solution['compliance'],
                    overall_score=best_solution['overall_score'],
                    trade_offs=best_solution['trade_offs'],
                    explanation=explanation
                )
                
                # Step 6: Record solution
                self._record_solution(final_solution)
                
                return final_solution
            
            def _generate_candidate_solutions(self):
                """Generate candidate hub configurations"""
                
                candidates = []
                
                # Generate different numbers of hubs
                for num_hubs in range(3, min(8, self.num_nodes)):
                    for _ in range(5):  # 5 candidates per hub count
                        # Use ethical constraints to guide selection
                        hub_combination = self._generate_ethical_hub_combination(num_hubs)
                        
                        # Calculate spoke assignments
                        spoke_assignments = self._calculate_spoke_assignments(hub_combination)
                        
                        candidates.append({
                            'hub_configuration': hub_combination,
                            'spoke_assignments': spoke_assignments
                        })
                
                return candidates
            
            def _generate_ethical_hub_combination(self, num_hubs):
                """Generate hub combination considering ethical factors"""
                
                # Calculate ethical scores for each potential hub
                hub_scores = []
                
                for i in range(self.num_nodes):
                    score = 0.0
                    
                    # Employment impact (higher is better)
                    employment_score = self.social_impact_factors[i]['employment_impact']
                    score += employment_score * 0.3
                    
                    # Community development need (higher is better)
                    community_score = self.community_data[i]['need_index']
                    score += community_score * 0.2
                    
                    # Wage equity (prefer moderate wages)
                    wage_score = 1.0 - abs(self.wage_levels[i] - np.mean(self.wage_levels)) / np.mean(self.wage_levels)
                    score += wage_score * 0.2
                    
                    # Environmental impact (lower emissions is better)
                    avg_emissions = np.mean([self.emission_factors[i][j] for j in range(self.num_nodes) if i != j])
                    emission_score = 1.0 - (avg_emissions / np.mean(self.emission_factors))
                    score += emission_score * 0.3
                    
                    hub_scores.append((i, score))
                
                # Sort by ethical score and select top candidates
                hub_scores.sort(key=lambda x: x[1], reverse=True)
                
                # Add some randomness for diversity
                top_candidates = hub_scores[:num_hubs * 2]
                selected_hubs = [h[0] for h in random.sample(top_candidates, num_hubs)]
                
                return selected_hubs
            
            def _calculate_spoke_assignments(self, hub_configuration):
                """Calculate optimal spoke assignments for given hub configuration"""
                
                assignments = {}
                
                for node in range(self.num_nodes):
                    if node not in hub_configuration:
                        # Find closest hub
                        min_cost = float('inf')
                        best_hub = None
                        
                        for hub in hub_configuration:
                            # Consider distance and ethical factors
                            distance = self.distance_matrix[node][hub]
                            
                            # Ethical preference for local hubs
                            ethical_bonus = 0.1 if self._get_node_region(node) == self._get_node_region(hub) else 0
                            
                            total_cost = distance * (1 - ethical_bonus)
                            
                            if total_cost < min_cost:
                                min_cost = total_cost
                                best_hub = hub
                        
                        assignments[node] = best_hub
                    else:
                        assignments[node] = node  # Hub assigned to itself
                
                return assignments
            
            def _calculate_stakeholder_utilities(self, solution):
                """Calculate utilities for all stakeholder groups"""
                
                utilities = {}
                
                # Calculate network performance metrics
                performance_metrics = self._calculate_network_performance(solution)
                
                for stakeholder_name, stakeholder in self.stakeholders.items():
                    utility = 0.0
                    
                    # Calculate utility based on stakeholder weights
                    for objective, weight in stakeholder.utility_weights.items():
                        metric_value = performance_metrics.get(objective, 0)
                        utility += metric_value * weight
                    
                    utilities[stakeholder_name] = utility
                
                return utilities
            
            def _calculate_network_performance(self, solution):
                """Calculate network performance metrics"""
                
                hub_configuration = solution['hub_configuration']
                spoke_assignments = solution['spoke_assignments']
                
                metrics = {}
                
                # Calculate total costs
                fixed_cost = sum(self.hub_costs[hub] for hub in hub_configuration)
                
                variable_cost = 0
                for i in range(self.num_nodes):
                    for j in range(self.num_nodes):
                        if i != j:
                            hub_i = spoke_assignments[i]
                            hub_j = spoke_assignments[j]
                            
                            cost = self.demand_matrix[i][j] * (
                                self.distance_matrix[i][hub_i] + 
                                0.75 * self.distance_matrix[hub_i][hub_j] +
                                0.75 * self.distance_matrix[hub_j][j]
                            )
                            variable_cost += cost
                
                total_cost = fixed_cost + variable_cost
                
                # Calculate service quality
                max_distance = np.max(self.distance_matrix[self.distance_matrix > 0])
                service_quality = 0.0
                total_flow = 0
                
                for i in range(self.num_nodes):
                    for j in range(self.num_nodes):
                        if i != j:
                            hub_i = spoke_assignments[i]
                            hub_j = spoke_assignments[j]
                            
                            distance = self.distance_matrix[i][hub_i] + \
                                         0.75 * self.distance_matrix[hub_i][hub_j] + \
                                         0.75 * self.distance_matrix[hub_j][j]
                            
                            service_quality += self.demand_matrix[i][j] * (1 - distance / max_distance)
                            total_flow += self.demand_matrix[i][j]
                
                service_quality = service_quality / total_flow if total_flow > 0 else 0
                
                # Calculate carbon emissions
                carbon_emissions = 0
                for i in range(self.num_nodes):
                    for j in range(self.num_nodes):
                        if i != j:
                            hub_i = spoke_assignments[i]
                            hub_j = spoke_assignments[j]
                            
                            emissions = self.demand_matrix[i][j] * (
                                self.emission_factors[i][hub_i] +
                                0.75 * self.emission_factors[hub_i][hub_j] +
                                0.75 * self.emission_factors[hub_j][j]
                            )
                            carbon_emissions += emissions
                
                # Calculate employment impact
                total_employment = 0
                for hub in hub_configuration:
                    employment_impact = self.social_impact_factors[hub]['employment_impact']
                    total_employment += employment_impact * self.capacity_limits[hub] / 10000
                
                # Calculate community impact
                community_impact = 0
                for hub in hub_configuration:
                    need_index = self.community_data[hub]['need_index']
                    community_impact += need_index * self.capacity_limits[hub] / 10000
                
                # Calculate wage equity
                hub_wages = [self.wage_levels[hub] for hub in hub_configuration]
                wage_std = np.std(hub_wages)
                wage_mean = np.mean(hub_wages)
                wage_equity = 1.0 - (wage_std / wage_mean) if wage_mean > 0 else 0
                
                # Calculate local employment
                local_employment = 0
                for hub in hub_configuration:
                    region = self._get_node_region(hub)
                    local_jobs = sum(1 for node in range(self.num_nodes) 
                                   if spoke_assignments[node] == hub and self._get_node_region(node) == region)
                    total_jobs = sum(1 for node in range(self.num_nodes) if spoke_assignments[node] == hub)
                    
                    if total_jobs > 0:
                        local_employment += local_jobs / total_jobs
                
                local_employment /= len(hub_configuration) if hub_configuration else 1
                
                # Calculate community investment
                community_investment = total_cost * 0.02  # 2% of total cost
                
                # Calculate normalized metrics
                metrics = {
                    'cost_efficiency': 1.0 - (total_cost / 10000000),  # Normalize to $10M
                    'revenue_growth': service_quality,  # Use service quality as proxy
                    'market_expansion': len(hub_configuration) / self.num_nodes,
                    'risk_management': wage_equity,
                    'employment_stability': total_employment / 100,
                    'wage_fairness': wage_equity,
                    'work_life_balance': 0.8,  # Simplified
                    'safety_standards': 0.9,  # Simplified
                    'training_opportunities': 0.7,  # Simplified
                    'service_quality': service_quality,
                    'price_competitiveness': 1.0 - (total_cost / 10000000),
                    'delivery_reliability': service_quality,
                    'product_quality': 0.9,  # Simplified
                    'environmental_impact': 1.0 - (carbon_emissions / 50000),
                    'carbon_emissions': 1.0 - (carbon_emissions / 50000),
                    'waste_generation': 0.8,  # Simplified
                    'energy_efficiency': 0.8,  # Simplified
                    'resource_conservation': 0.7,  # Simplified
                    'ecosystem_protection': 0.8,  # Simplified
                    'local_jobs': local_employment,
                    'economic_contribution': community_impact / 50,
                    'social_welfare': community_impact / 50,
                    'infrastructure_development': community_impact / 50,
                    'cultural_preservation': 0.8  # Simplified
                }
                
                return metrics
            
            def _evaluate_ethical_compliance(self, solution):
                """Evaluate ethical compliance of solution"""
                
                compliance = {}
                total_penalty = 0.0
                
                # Evaluate each constraint
                for constraint in self.ethical_constraints:
                    constraint_value = self._evaluate_constraint(solution, constraint)
                    
                    # Calculate penalty
                    if constraint.type == 'hard':
                        if constraint_value < constraint.threshold:
                            penalty = (constraint.threshold - constraint_value) * constraint.penalty_factor
                        else:
                            penalty = 0.0
                    elif constraint.type == 'soft':
                        if constraint_value < constraint.threshold:
                            penalty = (constraint.threshold - constraint_value) * constraint.penalty_factor * 0.5
                        else:
                            penalty = 0.0
                    else:  # preference
                        if constraint_value < constraint.threshold:
                            penalty = (constraint.threshold - constraint_value) * constraint.penalty_factor * 0.25
                        else:
                            penalty = 0.0
                    
                    compliance[constraint.name] = {
                        'value': constraint_value,
                        'threshold': constraint.threshold,
                        'compliance': constraint_value >= constraint.threshold,
                        'penalty': penalty
                    }
                    
                    total_penalty += penalty
                
                # Calculate overall compliance
                compliant_constraints = sum(1 for c in compliance.values() if c['compliance'])
                total_constraints = len(compliance)
                overall_compliance = compliant_constraints / total_constraints
                
                compliance['overall_compliance'] = overall_compliance
                compliance['total_penalty'] = total_penalty
                
                return compliance
            
            def _evaluate_constraint(self, solution, constraint):
                """Evaluate a specific constraint"""
                
                if constraint.name == 'maximum_carbon_emissions':
                    return self._calculate_total_emissions(solution)
                elif constraint.name == 'minimum_service_level':
                    return self._calculate_service_level(solution)
                elif constraint.name == 'fair_wage_distribution':
                    return self._calculate_wage_equity(solution)
                elif constraint.name == 'local_employment_minimum':
                    return self._calculate_local_employment(solution)
                elif constraint.name == 'community_investment_minimum':
                    return self._calculate_community_investment(solution)
                else:
                    return 1.0  # Default compliance
            
            def _calculate_total_emissions(self, solution):
                """Calculate total carbon emissions"""
                
                hub_configuration = solution['hub_configuration']
                spoke_assignments = solution['spoke_assignments']
                
                total_emissions = 0
                for i in range(self.num_nodes):
                    for j in range(self.num_nodes):
                        if i != j:
                            hub_i = spoke_assignments[i]
                            hub_j = spoke_assignments[j]
                            
                            emissions = self.demand_matrix[i][j] * (
                                self.emission_factors[i][hub_i] +
                                0.75 * self.emission_factors[hub_i][hub_j] +
                                0.75 * self.emission_factors[hub_j][j]
                            )
                            total_emissions += emissions
                
                return total_emissions
            
            def _calculate_service_level(self, solution):
                """Calculate service level"""
                
                performance = self._calculate_network_performance(solution)
                return performance['service_quality']
            
            def _calculate_wage_equity(self, solution):
                """Calculate wage equity"""
                
                hub_configuration = solution['hub_configuration']
                hub_wages = [self.wage_levels[hub] for hub in hub_configuration]
                
                if len(hub_wages) > 1:
                    wage_std = np.std(hub_wages)
                    wage_mean = np.mean(hub_wages)
                    return 1.0 - (wage_std / wage_mean) if wage_mean > 0 else 0
                else:
                    return 1.0
            
            def _calculate_local_employment(self, solution):
                """Calculate local employment percentage"""
                
                spoke_assignments = solution['spoke_assignments']
                local_employment = 0
                total_hubs = 0
                
                for hub in solution['hub_configuration']:
                    region = self._get_node_region(hub)
                    local_jobs = sum(1 for node in range(self.num_nodes) 
                                   if spoke_assignments[node] == hub and self._get_node_region(node) == region)
                    total_jobs = sum(1 for node in range(self.num_nodes) if spoke_assignments[node] == hub)
                    
                    if total_jobs > 0:
                        local_employment += local_jobs / total_jobs
                    total_hubs += 1
                
                return local_employment / total_hubs if total_hubs > 0 else 0
            
            def _calculate_community_investment(self, solution):
                """Calculate community investment percentage"""
                
                # Simplified: assume 2% of total cost goes to community investment
                return 0.02
            
            def _calculate_overall_score(self, utilities, compliance):
                """Calculate overall score combining utilities and compliance"""
                
                # Weighted stakeholder utility
                total_utility = 0.0
                for stakeholder_name, utility in utilities.items():
                    stakeholder = self.stakeholders[stakeholder_name]
                    total_utility += utility * stakeholder.priority
                
                # Compliance penalty
                compliance_penalty = compliance['total_penalty']
                
                # Overall score
                overall_score = total_utility - compliance_penalty
                
                return max(0, overall_score)
            
            def _analyze_trade_offs(self, utilities):
                """Analyze trade-offs between stakeholder interests"""
                
                trade_offs = {}
                
                # Calculate variance in utilities
                utility_values = list(utilities.values())
                utility_variance = np.var(utility_values)
                
                # Identify stakeholder conflicts
                max_utility = max(utilities.values())
                min_utility = min(utilities.values())
                utility_gap = max_utility - min_utility
                
                trade_offs['utility_variance'] = utility_variance
                trade_offs['utility_gap'] = utility_gap
                trade_offs['equity_score'] = 1.0 - (utility_gap / max_utility) if max_utility > 0 else 0
                
                return trade_offs
            
            def _generate_solution_explanation(self, best_solution):
                """Generate explanation for the selected solution"""
                
                explanation_parts = []
                
                # Hub selection explanation
                hub_config = best_solution['solution']['hub_configuration']
                explanation_parts.append(f"Selected {len(hub_config)} hubs to balance stakeholder interests")
                
                # Stakeholder utility explanation
                utilities = best_solution['utilities']
                best_stakeholder = max(utilities, key=utilities.get)
                worst_stakeholder = min(utilities, key=utilities.get)
                
                explanation_parts.append(f"Best utility for {best_stakeholder}: {utilities[best_stakeholder]:.3f}")
                explanation_parts.append(f"Lowest utility for {worst_stakeholder}: {utilities[worst_stakeholder]:.3f}")
                
                # Ethical compliance explanation
                compliance = best_solution['compliance']
                compliant_constraints = sum(1 for c in compliance.values() if isinstance(c, dict) and c.get('compliance', False))
                total_constraints = len([c for c in compliance.values() if isinstance(c, dict)])
                
                explanation_parts.append(f"Ethical compliance: {compliant_constraints}/{total_constraints} constraints satisfied")
                
                # Trade-off explanation
                trade_offs = best_solution['trade_offs']
                explanation_parts.append(f"Stakeholder equity score: {trade_offs['equity_score']:.3f}")
                
                return "; ".join(explanation_parts)
            
            def _record_solution(self, solution):
                """Record solution for analysis"""
                
                self.solution_history.append({
                    'timestamp': datetime.now(),
                    'solution': solution,
                    'hub_count': len(solution.hub_configuration),
                    'overall_score': solution.overall_score,
                    'ethical_compliance': solution.ethical_compliance['overall_compliance']
                })
                
                self.compliance_history.append({
                    'timestamp': datetime.now(),
                    'compliance': solution.ethical_compliance,
                    'utilities': solution.stakeholder_utilities,
                    'trade_offs': solution.trade_offs
                })
        
        # Supporting classes for the Ethical Framework
        class EthicalOptimizationEngine:
            """Optimization engine with ethical constraints"""
            
            def __init__(self):
                pass  # Would contain optimization algorithms
        
        class ImpactAssessmentSystem:
            """System for assessing social and environmental impacts"""
            
            def __init__(self):
                pass  # Would contain impact assessment methods
        
        class EthicalComplianceMonitor:
            """System for monitoring ethical compliance"""
            
            def __init__(self):
                pass  # Would contain compliance monitoring methods
        
        # Initialize the Value-Aligned & Ethical Framework
        print("Initializing Value-Aligned & Ethical Framework...")
        
        ethical_framework = ValueAlignedEthicalFramework(num_nodes=20)
        
        print(f"Framework Configuration:")
        print(f"- Network Size: {ethical_framework.num_nodes} nodes")
        print(f"- Stakeholder Groups: {len(ethical_framework.stakeholders)}")
        print(f"- Ethical Principles: {len(ethical_framework.ethical_principles)}")
        print(f"- Ethical Constraints: {len(ethical_framework.ethical_constraints)}")
        
        print(f"\nStakeholder Groups:")
        for name, stakeholder in ethical_framework.stakeholders.items():
            print(f"  - {name}: Priority {stakeholder.priority:.2f}, {len(stakeholder.interests)} interests")
        
        print(f"\nEthical Principles:")
        for name, principle in ethical_framework.ethical_principles.items():
            print(f"  - {name}: Threshold {principle.minimum_threshold:.1f}, Penalty {principle.penalty_weight:.2f}")
    except Exception as e:
        print(f'Error in cell: {e}')
except Exception as e:
    print(f'Error in cell: {e}')
```
