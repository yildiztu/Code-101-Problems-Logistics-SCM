### Key assumptions

- Multi-system integration: ERP, SCM, IoT sensors, and market data feeds
- Predictive analytics for demand forecasting and supplier risk assessment
- What-if scenario analysis for strategic sourcing decisions
- Continuous learning and adaptation from operational data

### Approach (step-by-step)
1. **Digital Twin Architecture**: Design 4-layer system (Physical → Connectivity → Data → Application)
2. **Real-Time Data Integration**: Connect IoT sensors, ERP systems, and market feeds
3. **Predictive Analytics**: Implement ML models for demand forecasting and risk prediction
4. **Scenario Analysis**: Create what-if simulations for sourcing strategy changes
5. **Decision Support**: Provide real-time recommendations and impact analysis
6. **Continuous Learning**: Update models based on actual performance data
7. **KPI Monitoring**: Track comprehensive performance metrics and alerts

### What to look for in the results
- Real-time monitoring dashboard with live supplier performance data
- Predictive analytics accuracy and confidence intervals
- What-if scenario analysis showing impact of sourcing decisions
- System integration metrics and data quality indicators
- Continuous learning improvements over time

```python
# Import required libraries for Integrated Digital Twin
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass, field
from typing import List, Dict, Tuple, Optional, Union, Any
import random
import time
import warnings
from collections import defaultdict, deque
from datetime import datetime, timedelta
import copy
import asyncio
from concurrent.futures import ThreadPoolExecutor
warnings.filterwarnings('ignore')

# Set style for better visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
@dataclass
class IoTDataPoint:
    """IoT sensor data point"""
    sensor_id: str
    timestamp: datetime
    value: float
    unit: str
    location: str
    quality_score: float  # 0-1 scale

@dataclass
class SupplierPerformance:
    """Real-time supplier performance metrics"""
    supplier_id: str
    timestamp: datetime
    on_time_delivery_rate: float
    quality_score: float
    cost_performance: float
    inventory_turnover: float
    disruption_risk_score: float
    capacity_utilization: float
    response_time: float  # hours

@dataclass
class MarketCondition:
    """Market condition data"""
    timestamp: datetime
    market_index: float
    commodity_price_index: float
    shipping_cost_index: float
    currency_exchange_rate: float
    geopolitical_risk_score: float
    demand_volatility: float

@dataclass
class ComponentDemand:
    """Component demand forecast"""
    component_id: str
    timestamp: datetime
    current_demand: float
    forecast_7_days: float
    forecast_30_days: float
    forecast_confidence: float
    safety_stock_level: float
    reorder_point: float

@dataclass
class SourcingDecision:
    """Sourcing decision with impact analysis"""
    decision_id: str
    timestamp: datetime
    component_id: str
    action_type: str  # 'single_sourcing', 'dual_sourcing', 'switch_supplier'
    selected_suppliers: List[str]
    allocation_ratios: List[float]
    expected_cost_impact: float
    expected_service_impact: float
    expected_risk_impact: float
    confidence_score: float
    implementation_time: timedelta

@dataclass
class DigitalTwinMetrics:
    """Digital twin performance metrics"""
    timestamp: datetime
    data_quality_score: float
    prediction_accuracy: float
    system_uptime: float
    response_time_ms: float
    synchronization_delay_ms: float
    active_sensors: int
    data_points_processed: int
    alerts_generated: int
    decisions_supported: int
```

```python
class DigitalTwinSimulation:
    """Integrated Digital Twin for Single vs. Dual Sourcing"""
    
    def __init__(self, components: List[Dict], suppliers: List[Dict]):
        self.components = components
        self.suppliers = suppliers
        
        # Digital twin state
        self.current_time = datetime.now()
        self.iot_data_stream = deque(maxlen=10000)
        self.supplier_performance = defaultdict(list)
        self.market_conditions = deque(maxlen=1000)
        self.component_demands = defaultdict(list)
        self.sourcing_decisions = deque(maxlen=1000)
        self.dt_metrics = deque(maxlen=1000)
        
        # System components
        self.iot_sensors = self._initialize_iot_sensors()
        self.predictive_models = self._initialize_predictive_models()
        self.scenario_engine = ScenarioEngine()
        self.alert_system = AlertSystem()
        
        # Performance tracking
        self.start_time = datetime.now()
        self.total_data_points = 0
        self.total_decisions = 0
        self.total_alerts = 0
    
    def _initialize_iot_sensors(self) -> Dict[str, Dict]:
        """Initialize IoT sensors for real-time monitoring"""
        
        sensors = {}
        
        # Supplier performance sensors
        for supplier in self.suppliers:
            sensors[f"supplier_{supplier['id']}_delivery"] = {
                'type': 'delivery_performance',
                'location': supplier['location'],
                'sampling_rate': 1.0,  # Hz
                'unit': 'percentage'
            }
            sensors[f"supplier_{supplier['id']}_quality"] = {
                'type': 'quality_score',
                'location': supplier['location'],
                'sampling_rate': 0.1,  # Hz
                'unit': 'score'
            }
            sensors[f"supplier_{supplier['id']}_capacity"] = {
                'type': 'capacity_utilization',
                'location': supplier['location'],
                'sampling_rate': 0.5,  # Hz
                'unit': 'percentage'
            }
        
        # Component inventory sensors
        for component in self.components:
            sensors[f"component_{component['id']}_inventory"] = {
                'type': 'inventory_level',
                'location': 'warehouse',
                'sampling_rate': 2.0,  # Hz
                'unit': 'units'
            }
            sensors[f"component_{component['id']}_demand"] = {
                'type': 'demand_rate',
                'location': 'warehouse',
                'sampling_rate': 1.0,  # Hz
                'unit': 'units_per_hour'
            }
        
        # Market condition sensors
        sensors['market_index'] = {
            'type': 'market_condition',
            'location': 'global',
            'sampling_rate': 0.1,  # Hz
            'unit': 'index'
        }
        sensors['commodity_prices'] = {
            'type': 'commodity_price',
            'location': 'global',
            'sampling_rate': 0.05,  # Hz
            'unit': 'index'
        }
        
        return sensors
    
    def _initialize_predictive_models(self) -> Dict[str, Any]:
        """Initialize predictive analytics models"""
        
        models = {
            'demand_forecast': DemandForecastModel(),
            'supplier_risk': SupplierRiskModel(),
            'cost_prediction': CostPredictionModel(),
            'disruption_prediction': DisruptionPredictionModel()
        }
        
        return models
    
    def simulate_real_time_data(self, duration_minutes: int = 60):
        """Simulate real-time data streaming"""
        
        print(f"Starting real-time simulation for {duration_minutes} minutes...")
        
        for minute in range(duration_minutes):
            self.current_time = datetime.now() + timedelta(minutes=minute)
            
            # Generate IoT sensor data
            iot_data = self._generate_iot_data()
            self.iot_data_stream.extend(iot_data)
            self.total_data_points += len(iot_data)
            
            # Update supplier performance
            supplier_perf = self._update_supplier_performance()
            for perf in supplier_perf:
                self.supplier_performance[perf.supplier_id].append(perf)
            
            # Update market conditions
            market_cond = self._update_market_conditions()
            self.market_conditions.append(market_cond)
            
            # Update component demands
            component_demand = self._update_component_demands()
            for demand in component_demand:
                self.component_demands[demand.component_id].append(demand)
            
            # Run predictive analytics
            predictions = self._run_predictive_analytics()
            
            # Generate sourcing decisions
            decisions = self._generate_sourcing_decisions(predictions)
            for decision in decisions:
                self.sourcing_decisions.append(decision)
                self.total_decisions += 1
            
            # Check for alerts
            alerts = self._check_alerts()
            self.total_alerts += len(alerts)
            
            # Update digital twin metrics
            dt_metrics = self._calculate_dt_metrics()
            self.dt_metrics.append(dt_metrics)
            
            # Progress update
            if minute % 10 == 0:
                print(f"  Minute {minute}: {len(iot_data)} data points, {len(decisions)} decisions, {len(alerts)} alerts")
        
        print(f"Simulation completed: {self.total_data_points} total data points processed")
    
    def _generate_iot_data(self) -> List[IoTDataPoint]:
        """Generate IoT sensor data points"""
        
        data_points = []
        
        for sensor_id, sensor_config in self.iot_sensors.items():
            # Generate data based on sensor type
            if sensor_config['type'] == 'delivery_performance':
                base_value = 0.85 + np.random.normal(0, 0.1)
                value = np.clip(base_value, 0.5, 1.0)
            elif sensor_config['type'] == 'quality_score':
                base_value = 0.90 + np.random.normal(0, 0.05)
                value = np.clip(base_value, 0.7, 1.0)
            elif sensor_config['type'] == 'capacity_utilization':
                base_value = 0.75 + np.random.normal(0, 0.15)
                value = np.clip(base_value, 0.3, 1.0)
            elif sensor_config['type'] == 'inventory_level':
                base_value = np.random.normal(500, 100)
                value = max(0, base_value)
            elif sensor_config['type'] == 'demand_rate':
                base_value = np.random.normal(20, 5)
                value = max(0, base_value)
            elif sensor_config['type'] == 'market_condition':
                base_value = np.random.normal(0, 0.3)
                value = base_value
            elif sensor_config['type'] == 'commodity_price':
                base_value = np.random.normal(100, 10)
                value = max(0, base_value)
            else:
                value = np.random.normal(0, 1)
            
            data_point = IoTDataPoint(
                sensor_id=sensor_id,
                timestamp=self.current_time,
                value=value,
                unit=sensor_config['unit'],
                location=sensor_config['location'],
                quality_score=0.9 + np.random.normal(0, 0.1)  # 90% average quality
            )
            
            data_points.append(data_point)
        
        return data_points
    
    def _update_supplier_performance(self) -> List[SupplierPerformance]:
        """Update supplier performance metrics"""
        
        performances = []
        
        for supplier in self.suppliers:
            # Get recent IoT data for this supplier
            supplier_data = [dp for dp in self.iot_data_stream 
                           if supplier['id'] in dp.sensor_id]
            
            if supplier_data:
                # Calculate performance metrics
                delivery_data = [dp for dp in supplier_data if 'delivery' in dp.sensor_id]
                quality_data = [dp for dp in supplier_data if 'quality' in dp.sensor_id]
                capacity_data = [dp for dp in supplier_data if 'capacity' in dp.sensor_id]
                
                on_time_delivery = np.mean([dp.value for dp in delivery_data]) if delivery_data else 0.85
                quality_score = np.mean([dp.value for dp in quality_data]) if quality_data else 0.90
                capacity_util = np.mean([dp.value for dp in capacity_data]) if capacity_data else 0.75
                
                performance = SupplierPerformance(
                    supplier_id=supplier['id'],
                    timestamp=self.current_time,
                    on_time_delivery_rate=on_time_delivery,
                    quality_score=quality_score,
                    cost_performance=supplier['unit_cost'],
                    inventory_turnover=np.random.normal(8, 2),
                    disruption_risk_score=max(0, 1.0 - on_time_delivery),
                    capacity_utilization=capacity_util,
                    response_time=np.random.normal(24, 6)  # hours
                )
                
                performances.append(performance)
        
        return performances
    
    def _update_market_conditions(self) -> MarketCondition:
        """Update market condition data"""
        
        market_data = [dp for dp in self.iot_data_stream 
                      if dp.sensor_id in ['market_index', 'commodity_prices']]
        
        market_index = np.mean([dp.value for dp in market_data 
                               if dp.sensor_id == 'market_index']) if market_data else 0.0
        commodity_prices = np.mean([dp.value for dp in market_data 
                                   if dp.sensor_id == 'commodity_prices']) if market_data else 100.0
        
        return MarketCondition(
            timestamp=self.current_time,
            market_index=market_index,
            commodity_price_index=commodity_prices,
            shipping_cost_index=100 + np.random.normal(0, 15),
            currency_exchange_rate=1.0 + np.random.normal(0, 0.05),
            geopolitical_risk_score=max(0, np.random.normal(0.2, 0.1)),
            demand_volatility=0.3 + np.random.normal(0, 0.1)
        )
    
    def _update_component_demands(self) -> List[ComponentDemand]:
        """Update component demand forecasts"""
        
        demands = []
        
        for component in self.components:
            # Get recent demand data
            demand_data = [dp for dp in self.iot_data_stream 
                          if component['id'] in dp.sensor_id and 'demand' in dp.sensor_id]
            
            current_demand = np.mean([dp.value for dp in demand_data]) if demand_data else 20.0
            
            # Use predictive model for forecasting
            forecast_7 = self.predictive_models['demand_forecast'].predict(
                component['id'], current_demand, 7
            )
            forecast_30 = self.predictive_models['demand_forecast'].predict(
                component['id'], current_demand, 30
            )
            
            demand = ComponentDemand(
                component_id=component['id'],
                timestamp=self.current_time,
                current_demand=current_demand,
                forecast_7_days=forecast_7,
                forecast_30_days=forecast_30,
                forecast_confidence=0.85 + np.random.normal(0, 0.1),
                safety_stock_level=component['safety_stock'],
                reorder_point=component['reorder_point']
            )
            
            demands.append(demand)
        
        return demands
    
    def _run_predictive_analytics(self) -> Dict[str, Any]:
        """Run predictive analytics models"""
        
        predictions = {}
        
        # Get recent market conditions
        recent_market = list(self.market_conditions)[-5:] if self.market_conditions else []
        
        # Supplier risk predictions
        for supplier in self.suppliers:
            supplier_perf = self.supplier_performance[supplier['id']]
            if supplier_perf:
                recent_perf = supplier_perf[-1]
                risk_score = self.predictive_models['supplier_risk'].predict(
                    recent_perf, recent_market
                )
                predictions[f"supplier_{supplier['id']}_risk"] = risk_score
        
        # Cost predictions
        for supplier in self.suppliers:
            cost_prediction = self.predictive_models['cost_prediction'].predict(
                supplier['id'], recent_market
            )
            predictions[f"supplier_{supplier['id']}_cost"] = cost_prediction
        
        # Disruption predictions
        disruption_risk = self.predictive_models['disruption_prediction'].predict(
            recent_market, self.suppliers
        )
        predictions['disruption_risk'] = disruption_risk
        
        return predictions
    
    def _generate_sourcing_decisions(self, predictions: Dict[str, Any]) -> List[SourcingDecision]:
        """Generate sourcing decisions based on predictions"""
        
        decisions = []
        
        for component in self.components:
            # Get component demand
            component_demand = self.component_demands[component['id']]
            if not component_demand:
                continue
            
            current_demand = component_demand[-1]
            
            # Analyze supplier options
            supplier_scores = []
            for supplier in self.suppliers:
                risk_score = predictions.get(f"supplier_{supplier['id']}_risk", 0.2)
                cost_prediction = predictions.get(f"supplier_{supplier['id']}_cost", supplier['unit_cost'])
                
                # Calculate composite score
                score = (
                    (1.0 - risk_score) * 0.4 +  # Risk weight
                    (1.0 / cost_prediction) * 0.3 +  # Cost weight (inverse)
                    supplier['reliability_score'] * 0.3  # Reliability weight
                )
                
                supplier_scores.append((supplier['id'], score, risk_score))
            
            # Sort by score
            supplier_scores.sort(key=lambda x: x[1], reverse=True)
            
            # Make sourcing decision
            if component['criticality_score'] > 0.8 and len(supplier_scores) >= 2:
                # Dual sourcing for critical components
                best_supplier = supplier_scores[0][0]
                second_supplier = supplier_scores[1][0]
                
                decision = SourcingDecision(
                    decision_id=f"decision_{len(self.sourcing_decisions)}",
                    timestamp=self.current_time,
                    component_id=component['id'],
                    action_type='dual_sourcing',
                    selected_suppliers=[best_supplier, second_supplier],
                    allocation_ratios=[0.7, 0.3],
                    expected_cost_impact=0.05,  # 5% cost increase
                    expected_service_impact=0.15,  # 15% service improvement
                    expected_risk_impact=-0.4,  # 40% risk reduction
                    confidence_score=0.85,
                    implementation_time=timedelta(days=14)
                )
            else:
                # Single sourcing
                best_supplier = supplier_scores[0][0]
                
                decision = SourcingDecision(
                    decision_id=f"decision_{len(self.sourcing_decisions)}",
                    timestamp=self.current_time,
                    component_id=component['id'],
                    action_type='single_sourcing',
                    selected_suppliers=[best_supplier],
                    allocation_ratios=[1.0],
                    expected_cost_impact=0.0,
                    expected_service_impact=0.0,
                    expected_risk_impact=0.0,
                    confidence_score=0.90,
                    implementation_time=timedelta(days=7)
                )
            
            decisions.append(decision)
        
        return decisions
    
    def _check_alerts(self) -> List[Dict[str, Any]]:
        """Check for alerts and warnings"""
        
        alerts = []
        
        # Check supplier performance alerts
        for supplier_id, performances in self.supplier_performance.items():
            if performances:
                latest_perf = performances[-1]
                
                if latest_perf.on_time_delivery_rate < 0.8:
                    alerts.append({
                        'type': 'supplier_performance',
                        'severity': 'high',
                        'message': f"Supplier {supplier_id} delivery rate below 80%",
                        'timestamp': self.current_time,
                        'value': latest_perf.on_time_delivery_rate
                    })
                
                if latest_perf.quality_score < 0.85:
                    alerts.append({
                        'type': 'quality_issue',
                        'severity': 'medium',
                        'message': f"Supplier {supplier_id} quality score below 85%",
                        'timestamp': self.current_time,
                        'value': latest_perf.quality_score
                    })
                
                if latest_perf.capacity_utilization > 0.9:
                    alerts.append({
                        'type': 'capacity_constraint',
                        'severity': 'high',
                        'message': f"Supplier {supplier_id} capacity utilization above 90%",
                        'timestamp': self.current_time,
                        'value': latest_perf.capacity_utilization
                    })
        
        # Check inventory alerts
        for component_id, demands in self.component_demands.items():
            if demands:
                latest_demand = demands[-1]
                
                # Get current inventory
                inventory_data = [dp for dp in self.iot_data_stream 
                                if component_id in dp.sensor_id and 'inventory' in dp.sensor_id]
                
                if inventory_data:
                    current_inventory = inventory_data[-1].value
                    
                    if current_inventory < latest_demand.reorder_point:
                        alerts.append({
                            'type': 'low_inventory',
                            'severity': 'high',
                            'message': f"Component {component_id} inventory below reorder point",
                            'timestamp': self.current_time,
                            'value': current_inventory
                        })
        
        # Check market condition alerts
        if self.market_conditions:
            latest_market = self.market_conditions[-1]
            
            if latest_market.market_index < -0.5:
                alerts.append({
                    'type': 'market_downturn',
                    'severity': 'medium',
                    'message': 'Market conditions indicate significant downturn',
                    'timestamp': self.current_time,
                    'value': latest_market.market_index
                })
            
            if latest_market.geopolitical_risk_score > 0.7:
                alerts.append({
                    'type': 'geopolitical_risk',
                    'severity': 'high',
                    'message': 'High geopolitical risk detected',
                    'timestamp': self.current_time,
                    'value': latest_market.geopolitical_risk_score
                })
        
        return alerts
    
    def _calculate_dt_metrics(self) -> DigitalTwinMetrics:
        """Calculate digital twin performance metrics"""
        
        # Calculate data quality score
        recent_data = list(self.iot_data_stream)[-100:] if len(self.iot_data_stream) >= 100 else list(self.iot_data_stream)
        data_quality = np.mean([dp.quality_score for dp in recent_data]) if recent_data else 0.9
        
        # Calculate prediction accuracy (simulated)
        prediction_accuracy = 0.85 + np.random.normal(0, 0.05)
        prediction_accuracy = np.clip(prediction_accuracy, 0.7, 0.95)
        
        # Calculate system uptime (simulated)
        uptime = 0.995 + np.random.normal(0, 0.005)
        uptime = np.clip(uptime, 0.98, 1.0)
        
        # Calculate response time
        response_time = 50 + np.random.normal(0, 20)  # milliseconds
        response_time = max(10, response_time)
        
        # Calculate synchronization delay
        sync_delay = 100 + np.random.normal(0, 50)  # milliseconds
        sync_delay = max(50, sync_delay)
        
        return DigitalTwinMetrics(
            timestamp=self.current_time,
            data_quality_score=data_quality,
            prediction_accuracy=prediction_accuracy,
            system_uptime=uptime,
            response_time_ms=response_time,
            synchronization_delay_ms=sync_delay,
            active_sensors=len(self.iot_sensors),
            data_points_processed=len(recent_data),
            alerts_generated=len(self._check_alerts()),
            decisions_supported=len(self._generate_sourcing_decisions(self._run_predictive_analytics()))
        )
    
    def run_scenario_analysis(self, scenario_name: str, scenario_params: Dict) -> Dict[str, Any]:
        """Run what-if scenario analysis"""
        
        print(f"Running scenario analysis: {scenario_name}")
        
        # Save current state
        original_state = copy.deepcopy(self.__dict__)
        
        try:
            # Apply scenario parameters
            self._apply_scenario_params(scenario_params)
            
            # Run simulation for scenario
            self.simulate_real_time_data(duration_minutes=30)
            
            # Collect results
            results = self._collect_scenario_results()
            
            return results
            
        finally:
            # Restore original state
            self.__dict__.update(original_state)
    
    def _apply_scenario_params(self, params: Dict):
        """Apply scenario parameters to the simulation"""
        
        if 'supplier_disruption' in params:
            # Simulate supplier disruption
            disrupted_supplier = params['supplier_disruption']
            for supplier in self.suppliers:
                if supplier['id'] == disrupted_supplier:
                    supplier['reliability_score'] *= 0.5
                    supplier['unit_cost'] *= 1.3
        
        if 'demand_surge' in params:
            # Simulate demand surge
            surge_factor = params['demand_surge']
            for component in self.components:
                component['demand_forecast'] *= surge_factor
        
        if 'cost_increase' in params:
            # Simulate cost increase
            cost_factor = params['cost_increase']
            for supplier in self.suppliers:
                supplier['unit_cost'] *= cost_factor
    
    def _collect_scenario_results(self) -> Dict[str, Any]:
        """Collect scenario analysis results"""
        
        results = {
            'total_decisions': len(self.sourcing_decisions),
            'total_alerts': self.total_alerts,
            'avg_prediction_accuracy': np.mean([m.prediction_accuracy for m in self.dt_metrics]) if self.dt_metrics else 0.0,
            'avg_data_quality': np.mean([m.data_quality_score for m in self.dt_metrics]) if self.dt_metrics else 0.0,
            'decision_types': defaultdict(int),
            'alert_types': defaultdict(int)
        }
        
        # Analyze decision types
        for decision in self.sourcing_decisions:
            results['decision_types'][decision.action_type] += 1
        
        # Analyze alert types
        for minute in range(30):
            alerts = self._check_alerts()
            for alert in alerts:
                results['alert_types'][alert['type']] += 1
        
        return results
    
    def generate_comprehensive_report(self) -> str:
        """Generate comprehensive digital twin report"""
        
        report = f"""
# Integrated Digital Twin Report
================================
Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
Simulation Duration: {(datetime.now() - self.start_time).total_seconds():.1f} seconds

## System Overview
- Total Components: {len(self.components)}
- Total Suppliers: {len(self.suppliers)}
- Active IoT Sensors: {len(self.iot_sensors)}
- Total Data Points Processed: {self.total_data_points:,}
- Total Sourcing Decisions: {self.total_decisions}
- Total Alerts Generated: {self.total_alerts}

## Performance Metrics
"""
        
        if self.dt_metrics:
            latest_metrics = self.dt_metrics[-1]
            report += f"""
- Data Quality Score: {latest_metrics.data_quality_score:.1%}
- Prediction Accuracy: {latest_metrics.prediction_accuracy:.1%}
- System Uptime: {latest_metrics.system_uptime:.1%}
- Response Time: {latest_metrics.response_time_ms:.1f} ms
- Synchronization Delay: {latest_metrics.synchronization_delay_ms:.1f} ms
"""
        
        report += f"""

## Supplier Performance Summary
"""
        
        for supplier_id, performances in self.supplier_performance.items():
            if performances:
                latest = performances[-1]
                report += f"""
- {supplier_id}:
  - On-Time Delivery: {latest.on_time_delivery_rate:.1%}
  - Quality Score: {latest.quality_score:.1%}
  - Capacity Utilization: {latest.capacity_utilization:.1%}
  - Disruption Risk: {latest.disruption_risk_score:.1%}
"""
        
        report += f"""

## Recent Sourcing Decisions
"""
        
        recent_decisions = list(self.sourcing_decisions)[-5:] if len(self.sourcing_decisions) >= 5 else list(self.sourcing_decisions)
        for decision in recent_decisions:
            report += f"""
- {decision.component_id}: {decision.action_type}
  - Suppliers: {', '.join(decision.selected_suppliers)}
  - Expected Cost Impact: {decision.expected_cost_impact:+.1%}
  - Expected Service Impact: {decision.expected_service_impact:+.1%}
  - Risk Mitigation: {decision.expected_risk_impact:+.1%}
  - Confidence: {decision.confidence_score:.1%}
"""
        
        report += f"""

## System Health
"""
        
        # Calculate system health score
        if self.dt_metrics:
            avg_data_quality = np.mean([m.data_quality_score for m in self.dt_metrics])
            avg_prediction_accuracy = np.mean([m.prediction_accuracy for m in self.dt_metrics])
            avg_uptime = np.mean([m.system_uptime for m in self.dt_metrics])
            
            health_score = (avg_data_quality + avg_prediction_accuracy + avg_uptime) / 3
            
            if health_score > 0.9:
                health_status = "EXCELLENT"
            elif health_score > 0.8:
                health_status = "GOOD"
            elif health_score > 0.7:
                health_status = "FAIR"
            else:
                health_status = "POOR"
            
            report += f"""
- Overall Health Score: {health_score:.1%} ({health_status})
- Data Quality: {avg_data_quality:.1%}
- Prediction Accuracy: {avg_prediction_accuracy:.1%}
- System Uptime: {avg_uptime:.1%}
"""
        
        report += f"""

## Recommendations
"""
        
        # Generate recommendations based on current state
        if self.supplier_performance:
            # Find suppliers with high risk
            high_risk_suppliers = []
            for supplier_id, performances in self.supplier_performance.items():
                if performances and performances[-1].disruption_risk_score > 0.6:
                    high_risk_suppliers.append(supplier_id)
            
            if high_risk_suppliers:
                report += f"- Consider dual sourcing for suppliers with high risk: {', '.join(high_risk_suppliers)}\n"
        
        if self.dt_metrics and self.dt_metrics[-1].prediction_accuracy < 0.8:
            report += "- Retrain predictive models to improve forecast accuracy\n"
        
        if self.total_alerts > 100:
            report += "- High alert volume indicates system stress - consider operational adjustments\n"
        
        report += """

================================
End of Report
"""
        
        return report
```

```python
class DemandForecastModel:
    """Simplified demand forecasting model"""
    
    def predict(self, component_id: str, current_demand: float, days_ahead: int) -> float:
        """Predict demand for future days"""
        
        # Simple exponential smoothing with trend
        alpha = 0.3  # Smoothing factor
        trend_factor = 1.0 + (days_ahead * 0.01)  # 1% growth per day
        
        # Add some randomness for realism
        noise = np.random.normal(0, 0.05)
        
        predicted_demand = current_demand * trend_factor * (1 + noise)
        return max(0, predicted_demand)


class SupplierRiskModel:
    """Supplier risk assessment model"""
    
    def predict(self, performance: SupplierPerformance, market_conditions: List[MarketCondition]) -> float:
        """Predict supplier risk score"""
        
        # Base risk from performance metrics
        performance_risk = 1.0 - (performance.on_time_delivery_rate * 0.4 + 
                                   performance.quality_score * 0.3 + 
                                   (1.0 - performance.capacity_utilization) * 0.3)
        
        # Market condition impact
        market_risk = 0.0
        if market_conditions:
            latest_market = market_conditions[-1]
            market_risk = (latest_market.geopolitical_risk_score * 0.5 +
                          abs(latest_market.market_index) * 0.3 +
                          latest_market.demand_volatility * 0.2)
        
        # Combined risk score
        total_risk = performance_risk * 0.7 + market_risk * 0.3
        return np.clip(total_risk, 0.0, 1.0)


class CostPredictionModel:
    """Cost prediction model"""
    
    def predict(self, supplier_id: str, market_conditions: List[MarketCondition]) -> float:
        """Predict future costs for supplier"""
        
        # Base cost (would come from supplier data)
        base_cost = 15.0 + hash(supplier_id) % 10  # Simulated base cost
        
        # Market adjustments
        market_adjustment = 1.0
        if market_conditions:
            latest_market = market_conditions[-1]
            market_adjustment = (1.0 + latest_market.commodity_price_index / 1000 +
                                latest_market.shipping_cost_index / 1000 +
                                latest_market.currency_exchange_rate - 1.0)
        
        predicted_cost = base_cost * market_adjustment
        return max(0, predicted_cost)


class DisruptionPredictionModel:
    """Disruption prediction model"""
    
    def predict(self, market_conditions: List[MarketCondition], suppliers: List[Dict]) -> float:
        """Predict overall disruption risk"""
        
        base_risk = 0.1  # Base disruption probability
        
        # Market condition impact
        market_impact = 0.0
        if market_conditions:
            latest_market = market_conditions[-1]
            market_impact = (latest_market.geopolitical_risk_score * 0.6 +
                          abs(latest_market.market_index) * 0.4)
        
        # Supplier concentration risk
        concentration_risk = 0.1 * (1.0 - len(suppliers) / 10.0)  # Fewer suppliers = higher risk
        
        total_risk = base_risk + market_impact + concentration_risk
        return np.clip(total_risk, 0.0, 1.0)


class ScenarioEngine:
    """Scenario analysis engine"""
    
    def __init__(self):
        self.scenarios = {
            'supplier_disruption': {
                'description': 'Major supplier disruption',
                'params': {'supplier_disruption': 'supplier_1'},
                'duration_minutes': 30
            },
            'demand_surge': {
                'description': 'Unexpected demand surge',
                'params': {'demand_surge': 1.5},
                'duration_minutes': 30
            },
            'cost_increase': {
                'description': 'Market cost increase',
                'params': {'cost_increase': 1.2},
                'duration_minutes': 30
            },
            'combined_crisis': {
                'description': 'Multiple simultaneous disruptions',
                'params': {
                    'supplier_disruption': 'supplier_1',
                    'demand_surge': 1.3,
                    'cost_increase': 1.15
                },
                'duration_minutes': 45
            }
        }
    
    def get_available_scenarios(self) -> Dict[str, Dict]:
        """Get available scenarios"""
        return self.scenarios


class AlertSystem:
    """Alert management system"""
    
    def __init__(self):
        self.alert_history = deque(maxlen=1000)
        self.alert_rules = {
            'critical': ['supplier_performance', 'capacity_constraint', 'low_inventory', 'geopolitical_risk'],
            'warning': ['quality_issue', 'market_downturn'],
            'info': ['system_performance']
        }
    
    def process_alert(self, alert: Dict[str, Any]):
        """Process and categorize alert"""
        
        alert['processed_at'] = datetime.now()
        alert['acknowledged'] = False
        
        # Categorize alert
        for category, alert_types in self.alert_rules.items():
            if alert['type'] in alert_types:
                alert['category'] = category
                break
        else:
            alert['category'] = 'info'
        
        self.alert_history.append(alert)
        
        return alert
    
    def get_alert_statistics(self) -> Dict[str, Any]:
        """Get alert statistics"""
        
        if not self.alert_history:
            return {
                'total_alerts': 0,
                'by_category': defaultdict(int),
                'by_type': defaultdict(int),
                'by_severity': defaultdict(int)
            }
        
        stats = {
            'total_alerts': len(self.alert_history),
            'by_category': defaultdict(int),
            'by_type': defaultdict(int),
            'by_severity': defaultdict(int)
        }
        
        for alert in self.alert_history:
            stats['by_category'][alert['category']] += 1
            stats['by_type'][alert['type']] += 1
            stats['by_severity'][alert['severity']] += 1
        
        return stats
```
