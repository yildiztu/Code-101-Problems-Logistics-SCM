```python
@dataclass
class VehicleType:
    """Represents a vehicle type with characteristics"""
    id: int
    name: str
    capacity: int  # units of cargo
    acquisition_cost: float  # one-time cost
    operating_costs: Dict[int, float]  # cost per route
    service_rate: float  # requests per day can serve
    carbon_emissions: float  # kg CO2 per km
    noise_level: float  # decibels
    safety_rating: float  # safety score (0-1)
    accessibility: float  # accessibility score (0-1)

@dataclass
class Route:
    """Represents a route with demand characteristics"""
    id: int
    name: str
    demand_units: int  # total demand units per day
    arrival_rate: float  # requests per day (Poisson)
    max_delay_prob: float = 0.1
    population_density: float = 1.0  # people per km²
    environmental_sensitivity: float = 1.0  # environmental impact factor
    social_equity_factor: float = 1.0  # social equity importance

@dataclass
class Stakeholder:
    """Represents a stakeholder group with priorities"""
    id: int
    name: str
    stakeholder_type: str  # 'customers', 'employees', 'community', 'environment', 'investors'
    priorities: Dict[str, float]  # priority weights for different objectives
    influence_level: float  # influence on decisions (0-1)
    vulnerability: float  # vulnerability to negative impacts (0-1)
    equity_concern: float  # concern for fairness (0-1)

@dataclass
class EthicalPrinciple:
    """Represents an ethical principle with constraints"""
    id: int
    name: str
    description: str
    constraints: List[str]  # specific constraints
    weight: float  # importance weight (0-1)
    violation_penalty: float  # penalty for violation (0-1)
    measurable: bool  # whether compliance can be measured

@dataclass
class EthicalDecision:
    """Represents an ethical decision with compliance analysis"""
    fleet_allocation: np.ndarray
    objectives: Dict[str, float]  # multi-objective scores
    ethical_compliance: Dict[str, float]  # compliance scores per principle
    stakeholder_impacts: Dict[int, Dict[str, float]]  # impacts per stakeholder
    equity_score: float  # overall equity score (0-1)
    sustainability_score: float  # sustainability score (0-1)
    transparency_score: float  # transparency score (0-1)
    total_ethical_score: float  # overall ethical score (0-1)
    reasoning: List[str]  # ethical reasoning

@dataclass
class EthicalResult:
    """Represents the result of ethical optimization"""
    optimal_fleet: Dict[Tuple[int, int], int]  # optimal allocation
    total_cost: float  # total cost of optimal solution
    multi_objective_scores: Dict[str, float]  # scores for all objectives
    ethical_compliance: Dict[str, float]  # compliance with ethical principles
    stakeholder_equity: Dict[int, float]  # equity scores per stakeholder
    sustainability_metrics: Dict[str, float]  # environmental and social metrics
    transparency_metrics: Dict[str, float]  # transparency and accountability metrics
    computation_time: float  # total optimization time
    insights: List[str]  # key insights from ethical analysis
```

```python
class ConstitutionalAIFramework:
    """Constitutional AI framework for ethical fleet optimization"""
    
    def __init__(self, vehicles: List[VehicleType], routes: List[Route]):
        self.vehicles = vehicles
        self.routes = routes
        self.num_vehicles = len(vehicles)
        self.num_routes = len(routes)
        
        # Initialize constitutional framework
        self.ethical_principles = self._initialize_ethical_principles()
        self.stakeholders = self._initialize_stakeholders()
        self.objectives = self._initialize_objectives()
        
        # Ethical optimization parameters
        self.ethical_constraints = self._define_ethical_constraints()
        self.equity_weights = self._define_equity_weights()
        self.sustainability_targets = self._define_sustainability_targets()
        
        print(f"⚖️ Constitutional AI Framework Initialized:")
        print(f"  Ethical Principles: {len(self.ethical_principles)}")
        print(f"  Stakeholders: {len(self.stakeholders)}")
        print(f"  Objectives: {len(self.objectives)}")
        print(f"  Ethical Constraints: {len(self.ethical_constraints)}")
    
    def _initialize_ethical_principles(self) -> List[EthicalPrinciple]:
        """Initialize core ethical principles"""
        return [
            EthicalPrinciple(
                id=1,
                name="Environmental Responsibility",
                description="Minimize environmental impact and promote sustainability",
                constraints=[
                    "Carbon emissions must not exceed threshold",
                    "Noise pollution must be minimized",
                    "Environmental sensitivity areas must be protected"
                ],
                weight=0.25,
                violation_penalty=0.8,
                measurable=True
            ),
            EthicalPrinciple(
                id=2,
                name="Social Equity",
                description="Ensure fair distribution of benefits and burdens",
                constraints=[
                    "All communities must receive adequate service",
                    "Vulnerable populations must be protected",
                    "Service quality must be equitable across areas"
                ],
                weight=0.25,
                violation_penalty=0.9,
                measurable=True
            ),
            EthicalPrinciple(
                id=3,
                name="Safety & Well-being",
                description="Prioritize human safety and well-being",
                constraints=[
                    "Vehicle safety standards must be maintained",
                    "Driver working conditions must be safe",
                    "Community safety must be preserved"
                ],
                weight=0.20,
                violation_penalty=0.95,
                measurable=True
            ),
            EthicalPrinciple(
                id=4,
                name="Economic Viability",
                description="Ensure economic sustainability and efficiency",
                constraints=[
                    "Costs must remain within reasonable bounds",
                    "Resource efficiency must be optimized",
                    "Long-term financial sustainability required"
                ],
                weight=0.15,
                violation_penalty=0.6,
                measurable=True
            ),
            EthicalPrinciple(
                id=5,
                name="Transparency & Accountability",
                description="Ensure decisions are transparent and accountable",
                constraints=[
                    "Decision reasoning must be explainable",
                    "Stakeholder input must be considered",
                    "Performance metrics must be publicly available"
                ],
                weight=0.15,
                violation_penalty=0.5,
                measurable=True
            )
        ]
    
    def _initialize_stakeholders(self) -> List[Stakeholder]:
        """Initialize stakeholder groups"""
        return [
            Stakeholder(
                id=1,
                name="Customers",
                stakeholder_type="customers",
                priorities={
                    "service_quality": 0.4,
                    "cost": 0.3,
                    "reliability": 0.2,
                    "accessibility": 0.1
                },
                influence_level=0.3,
                vulnerability=0.6,
                equity_concern=0.7
            ),
            Stakeholder(
                id=2,
                name="Employees",
                stakeholder_type="employees",
                priorities={
                    "safety": 0.4,
                    "wages": 0.3,
                    "working_conditions": 0.2,
                    "job_security": 0.1
                },
                influence_level=0.2,
                vulnerability=0.4,
                equity_concern=0.6
            ),
            Stakeholder(
                id=3,
                name="Local Communities",
                stakeholder_type="community",
                priorities={
                    "environmental_quality": 0.3,
                    "noise_reduction": 0.2,
                    "local_jobs": 0.2,
                    "community_benefits": 0.3
                },
                influence_level=0.2,
                vulnerability=0.8,
                equity_concern=0.9
            ),
            Stakeholder(
                id=4,
                name="Environment",
                stakeholder_type="environment",
                priorities={
                    "carbon_reduction": 0.4,
                    "biodiversity": 0.3,
                    "pollution_control": 0.2,
                    "sustainability": 0.1
                },
                influence_level=0.1,
                vulnerability=0.9,
                equity_concern=1.0
            ),
            Stakeholder(
                id=5,
                name="Investors",
                stakeholder_type="investors",
                priorities={
                    "profitability": 0.5,
                    "risk_management": 0.3,
                    "growth": 0.1,
                    "sustainability": 0.1
                },
                influence_level=0.2,
                vulnerability=0.2,
                equity_concern=0.3
            )
        ]
    
    def _initialize_objectives(self) -> Dict[str, Dict]:
        """Initialize multi-objective optimization goals"""
        return {
            "economic": {
                "name": "Economic Efficiency",
                "description": "Minimize total cost while maintaining service quality",
                "direction": "minimize",
                "weight": 0.3
            },
            "environmental": {
                "name": "Environmental Sustainability",
                "description": "Minimize environmental impact and carbon footprint",
                "direction": "minimize",
                "weight": 0.25
            },
            "social": {
                "name": "Social Responsibility",
                "description": "Maximize social benefits and equity",
                "direction": "maximize",
                "weight": 0.25
            },
            "safety": {
                "name": "Safety & Well-being",
                "description": "Maximize safety and minimize risks",
                "direction": "maximize",
                "weight": 0.2
            }
        }
    
    def _define_ethical_constraints(self) -> Dict[str, Any]:
        """Define ethical constraints for optimization"""
        return {
            "max_carbon_per_vehicle": 50.0,  # kg CO2 per km per vehicle
            "max_noise_level": 70.0,  # decibels
            "min_service_equity": 0.8,  # minimum service equity score
            "max_cost_increase": 0.2,  # maximum cost increase vs optimal
            "min_safety_score": 0.8,  # minimum average safety score
            "max_vulnerability_impact": 0.3  # maximum impact on vulnerable stakeholders
        }
    
    def _define_equity_weights(self) -> Dict[str, float]:
        """Define equity weights for stakeholder considerations"""
        return {
            "vulnerability_multiplier": 2.0,  # weight for vulnerable stakeholders
            "equity_threshold": 0.8,  # minimum equity requirement
            "distribution_fairness": 0.7,  # fairness in benefit distribution
            "accessibility_requirement": 0.9  # accessibility requirement
        }
    
    def _define_sustainability_targets(self) -> Dict[str, float]:
        """Define sustainability targets and thresholds"""
        return {
            "carbon_reduction_target": 0.3,  # 30% reduction target
            "renewable_energy_ratio": 0.5,  # 50% renewable energy target
            "waste_reduction_target": 0.4,  # 40% waste reduction target
            "biodiversity_protection": 0.8  # 80% biodiversity protection target
        }
    
    def evaluate_ethical_compliance(self, fleet_allocation: np.ndarray) -> Dict[str, float]:
        """
        Evaluate compliance with ethical principles
        
        Args:
            fleet_allocation: vehicles x routes matrix
        
        Returns:
            Compliance scores per principle
        """
        compliance_scores = {}
        
        # Environmental Responsibility
        environmental_score = self._evaluate_environmental_compliance(fleet_allocation)
        compliance_scores["environmental"] = environmental_score
        
        # Social Equity
        equity_score = self._evaluate_social_equity_compliance(fleet_allocation)
        compliance_scores["social_equity"] = equity_score
        
        # Safety & Well-being
        safety_score = self._evaluate_safety_compliance(fleet_allocation)
        compliance_scores["safety"] = safety_score
        
        # Economic Viability
        economic_score = self._evaluate_economic_compliance(fleet_allocation)
        compliance_scores["economic"] = economic_score
        
        # Transparency & Accountability
        transparency_score = self._evaluate_transparency_compliance(fleet_allocation)
        compliance_scores["transparency"] = transparency_score
        
        return compliance_scores
    
    def _evaluate_environmental_compliance(self, fleet_allocation: np.ndarray) -> float:
        """Evaluate environmental compliance"""
        total_emissions = 0.0
        total_vehicles = np.sum(fleet_allocation)
        
        for i, vehicle in enumerate(self.vehicles):
            for j, route in enumerate(self.routes):
                count = fleet_allocation[i, j]
                if count > 0:
                    # Calculate emissions for this vehicle-route combination
                    emissions = count * vehicle.carbon_emissions * 100  # 100 km per route
                    total_emissions += emissions
        
        # Check against constraints
        max_allowed_emissions = total_vehicles * self.ethical_constraints["max_carbon_per_vehicle"] * 100
        
        if max_allowed_emissions > 0:
            compliance_score = max(0, 1 - (total_emissions - max_allowed_emissions) / max_allowed_emissions)
        else:
            compliance_score = 1.0
        
        return min(1.0, compliance_score)
    
    def _evaluate_social_equity_compliance(self, fleet_allocation: np.ndarray) -> float:
        """Evaluate social equity compliance"""
        service_levels = []
        
        for j, route in enumerate(self.routes):
            capacity = 0.0
            for i, vehicle in enumerate(self.vehicles):
                capacity += fleet_allocation[i, j] * vehicle.capacity
            
            service_level = min(1.0, capacity / route.demand_units) if route.demand_units > 0 else 1.0
            service_levels.append(service_level)
        
        # Calculate equity score (minimum service level across routes)
        equity_score = min(service_levels)
        
        # Apply equity threshold
        min_equity = self.ethical_constraints["min_service_equity"]
        if equity_score >= min_equity:
            return equity_score
        else:
            return equity_score / min_equity  # Penalty for not meeting threshold
    
    def _evaluate_safety_compliance(self, fleet_allocation: np.ndarray) -> float:
        """Evaluate safety compliance"""
        total_vehicles = np.sum(fleet_allocation)
        weighted_safety = 0.0
        
        for i, vehicle in enumerate(self.vehicles):
            for j, route in enumerate(self.routes):
                count = fleet_allocation[i, j]
                if count > 0:
                    weighted_safety += count * vehicle.safety_rating
        
        if total_vehicles > 0:
            avg_safety = weighted_safety / total_vehicles
        else:
            avg_safety = 1.0
        
        # Check against minimum safety requirement
        min_safety = self.ethical_constraints["min_safety_score"]
        
        if avg_safety >= min_safety:
            return avg_safety
        else:
            return avg_safety / min_safety  # Penalty for not meeting threshold
    
    def _evaluate_economic_compliance(self, fleet_allocation: np.ndarray) -> float:
        """Evaluate economic compliance"""
        total_cost = self._calculate_total_cost(fleet_allocation)
        
        # Compare with baseline cost (optimal without constraints)
        baseline_cost = 320000  # From previous tiers
        max_allowed_increase = self.ethical_constraints["max_cost_increase"]
        max_cost = baseline_cost * (1 + max_allowed_increase)
        
        if total_cost <= max_cost:
            return 1.0 - (total_cost - baseline_cost) / (max_cost - baseline_cost)
        else:
            return max(0, 1 - (total_cost - max_cost) / baseline_cost)  # Penalty for exceeding
    
    def _evaluate_transparency_compliance(self, fleet_allocation: np.ndarray) -> float:
        """Evaluate transparency compliance (simplified)"""
        # In practice, this would evaluate explainability, stakeholder input, etc.
        # For demonstration, we'll use a simplified approach
        
        # Base transparency score
        transparency_score = 0.8
        
        # Adjust based on decision complexity
        decision_complexity = np.sum(fleet_allocation)
        if decision_complexity > 10:
            transparency_score *= 0.9  # More complex decisions are harder to explain
        
        # Adjust based on stakeholder diversity
        stakeholder_diversity = len(self.stakeholders)
        if stakeholder_diversity > 4:
            transparency_score *= 0.95  # More stakeholders require more transparency
        
        return min(1.0, transparency_score)
    
    def _calculate_total_cost(self, fleet_allocation: np.ndarray) -> float:
        """Calculate total cost of fleet allocation"""
        total_cost = 0.0
        
        for i, vehicle in enumerate(self.vehicles):
            for j, route in enumerate(self.routes):
                count = fleet_allocation[i, j]
                if count > 0:
                    total_cost += count * (vehicle.acquisition_cost + vehicle.operating_costs[route.id])
        
        return total_cost
```

```python
class MultiObjectiveOptimizer:
    """Multi-objective optimizer with ethical constraints"""
    
    def __init__(self, constitutional_ai: ConstitutionalAIFramework):
        self.constitutional_ai = constitutional_ai
        self.vehicles = constitutional_ai.vehicles
        self.routes = constitutional_ai.routes
        self.num_vehicles = len(vehicles)
        self.num_routes = len(routes)
        
        # Optimization parameters
        self.objective_weights = {obj["name"]: obj["weight"] for obj in constitutional_ai.objectives.values()}
        self.constraint_weights = {p.name: p.weight for p in constitutional_ai.ethical_principles}
        
        print(f"🎯 Multi-Objective Optimizer Initialized:")
        print(f"  Objectives: {len(self.objective_weights)}")
        print(f"  Ethical Principles: {len(self.constraint_weights)}")
        print(f"  Optimization Method: Weighted sum with constraints")
    
    def optimize_fleet(self, context: Dict[str, Any] = None) -> EthicalDecision:
        """
        Optimize fleet allocation with ethical constraints
        
        Args:
            context: Optional context for optimization
        
        Returns:
            Ethical decision with full analysis
        """
        if context is None:
            context = self._create_default_context()
        
        # Generate candidate solutions
        candidates = self._generate_candidates(context)
        
        # Evaluate all candidates
        evaluated_candidates = []
        for candidate in candidates:
            evaluation = self._evaluate_candidate(candidate, context)
            evaluated_candidates.append(evaluation)
        
        # Select best solution
        best_decision = max(evaluated_candidates, key=lambda x: x.total_ethical_score)
        
        return best_decision
    
    def _create_default_context(self) -> Dict[str, Any]:
        """Create default optimization context"""
        return {
            "demands": np.array([15, 25]),
            "budget_limit": 500000,
            "environmental_priority": 0.8,
            "social_priority": 0.7,
            "safety_priority": 0.9,
            "equity_priority": 0.8
        }
    
    def _generate_candidates(self, context: Dict[str, Any]) -> List[np.ndarray]:
        """Generate candidate fleet allocations"""
        candidates = []
        
        # Generate diverse candidates
        base_allocations = [
            # Economic focus
            np.array([[3, 2], [1, 2], [0, 1]], dtype=float),
            # Environmental focus
            np.array([[4, 3], [0, 1], [0, 0]], dtype=float),
            # Social equity focus
            np.array([[2, 2], [2, 2], [1, 1]], dtype=float),
            # Safety focus
            np.array([[2, 2], [1, 2], [1, 1]], dtype=float),
            # Balanced approach
            np.array([[2, 2], [1, 2], [0, 1]], dtype=float)
        ]
        
        # Add variations
        for base in base_allocations:
            candidates.append(base)
            # Add small variations
            for _ in range(3):
                variation = base + np.random.normal(0, 0.3, base.shape)
                variation = np.maximum(variation, 0)  # Ensure non-negative
                candidates.append(variation)
        
        return candidates
    
    def _evaluate_candidate(self, fleet_allocation: np.ndarray, context: Dict[str, Any]) -> EthicalDecision:
        """Evaluate a candidate fleet allocation"""
        # Calculate multi-objective scores
        objectives = self._calculate_objectives(fleet_allocation, context)
        
        # Evaluate ethical compliance
        ethical_compliance = self.constitutional_ai.evaluate_ethical_compliance(fleet_allocation)
        
        # Calculate stakeholder impacts
        stakeholder_impacts = self._calculate_stakeholder_impacts(fleet_allocation)
        
        # Calculate equity score
        equity_score = self._calculate_equity_score(stakeholder_impacts)
        
        # Calculate sustainability score
        sustainability_score = self._calculate_sustainability_score(fleet_allocation, objectives)
        
        # Calculate transparency score
        transparency_score = self._calculate_transparency_score(fleet_allocation)
        
        # Calculate total ethical score
        total_ethical_score = self._calculate_total_ethical_score(
            objectives, ethical_compliance, equity_score, sustainability_score, transparency_score
        )
        
        # Generate reasoning
        reasoning = self._generate_ethical_reasoning(
            fleet_allocation, objectives, ethical_compliance, equity_score
        )
        
        return EthicalDecision(
            fleet_allocation=fleet_allocation,
            objectives=objectives,
            ethical_compliance=ethical_compliance,
            stakeholder_impacts=stakeholder_impacts,
            equity_score=equity_score,
            sustainability_score=sustainability_score,
            transparency_score=transparency_score,
            total_ethical_score=total_ethical_score,
            reasoning=reasoning
        )
    
    def _calculate_objectives(self, fleet_allocation: np.ndarray, context: Dict[str, Any]) -> Dict[str, float]:
        """Calculate multi-objective scores"""
        objectives = {}
        
        # Economic objective (minimize cost)
        total_cost = self.constitutional_ai._calculate_total_cost(fleet_allocation)
        objectives["economic"] = max(0, 1 - (total_cost - 320000) / 320000)  # Normalize around optimal
        
        # Environmental objective (minimize impact)
        total_emissions = 0.0
        for i, vehicle in enumerate(self.vehicles):
            for j, route in enumerate(self.routes):
                count = fleet_allocation[i, j]
                if count > 0:
                    total_emissions += count * vehicle.carbon_emissions * 100
        objectives["environmental"] = max(0, 1 - total_emissions / 10000)  # Normalize
        
        # Social objective (maximize service equity)
        service_levels = []
        for j, route in enumerate(self.routes):
            capacity = 0.0
            for i, vehicle in enumerate(self.vehicles):
                capacity += fleet_allocation[i, j] * vehicle.capacity
            service_level = min(1.0, capacity / route.demand_units) if route.demand_units > 0 else 1.0
            service_levels.append(service_level)
        objectives["social"] = min(service_levels)  # Use minimum service level as social metric
        
        # Safety objective (maximize safety)
        total_vehicles = np.sum(fleet_allocation)
        weighted_safety = 0.0
        for i, vehicle in enumerate(self.vehicles):
            for j, route in enumerate(self.routes):
                count = fleet_allocation[i, j]
                if count > 0:
                    weighted_safety += count * vehicle.safety_rating
        avg_safety = weighted_safety / total_vehicles if total_vehicles > 0 else 1.0
        objectives["safety"] = avg_safety
        
        return objectives
    
    def _calculate_stakeholder_impacts(self, fleet_allocation: np.ndarray) -> Dict[int, Dict[str, float]]:
        """Calculate impacts on each stakeholder group"""
        impacts = {}
        
        for stakeholder in self.constitutional_ai.stakeholders:
            stakeholder_impacts = {}
            
            # Calculate service impact
            service_levels = []
            for j, route in enumerate(self.routes):
                capacity = 0.0
                for i, vehicle in enumerate(self.vehicles):
                    capacity += fleet_allocation[i, j] * vehicle.capacity
                service_level = min(1.0, capacity / route.demand_units) if route.demand_units > 0 else 1.0
                service_levels.append(service_level)
            
            stakeholder_impacts["service"] = np.mean(service_levels)
            
            # Calculate cost impact (negative for cost-sensitive stakeholders)
            total_cost = self.constitutional_ai._calculate_total_cost(fleet_allocation)
            cost_impact = -total_cost / 500000  # Normalize
            stakeholder_impacts["cost"] = cost_impact * stakeholder.priorities.get("cost", 0)
            
            # Calculate environmental impact
            total_emissions = 0.0
            for i, vehicle in enumerate(self.vehicles):
                for j, route in enumerate(self.routes):
                    count = fleet_allocation[i, j]
                    if count > 0:
                        total_emissions += count * vehicle.carbon_emissions * 100
            env_impact = -total_emissions / 10000  # Negative for emissions
            stakeholder_impacts["environmental"] = env_impact * stakeholder.priorities.get("environmental_quality", 0)
            
            # Calculate overall impact
            overall_impact = (
                stakeholder_impacts["service"] * 0.4 +
                stakeholder_impacts["cost"] * 0.3 +
                stakeholder_impacts["environmental"] * 0.3
            )
            stakeholder_impacts["overall"] = overall_impact
            
            impacts[stakeholder.id] = stakeholder_impacts
        
        return impacts
    
    def _calculate_equity_score(self, stakeholder_impacts: Dict[int, Dict[str, float]]) -> float:
        """Calculate equity score across stakeholders"""
        overall_impacts = [impacts["overall"] for impacts in stakeholder_impacts.values()]
        
        # Calculate equity based on vulnerability weighting
        weighted_impacts = []
        for stakeholder_id, impacts in stakeholder_impacts.items():
            stakeholder = next(s for s in self.constitutional_ai.stakeholders if s.id == stakeholder_id)
            vulnerability_weight = 1 + stakeholder.vulnerability  # Higher weight for vulnerable
            weighted_impacts.append(impacts["overall"] * vulnerability_weight)
        
        # Equity score: balance between stakeholders
        if len(weighted_impacts) > 0:
            equity_score = 1 - np.std(weighted_impacts) / (np.mean(weighted_impacts) + 1e-6)
        else:
            equity_score = 1.0
        
        return max(0, min(1, equity_score))
    
    def _calculate_sustainability_score(self, fleet_allocation: np.ndarray, 
                                   objectives: Dict[str, float]) -> float:
        """Calculate sustainability score"""
        # Combine environmental and social aspects
        env_score = objectives.get("environmental", 0)
        social_score = objectives.get("social", 0)
        
        # Add long-term sustainability factors
        total_vehicles = np.sum(fleet_allocation)
        efficiency_bonus = min(1.0, 15 / total_vehicles) if total_vehicles > 0 else 1.0  # Fewer vehicles = more efficient
        
        sustainability_score = (env_score * 0.4 + social_score * 0.4 + efficiency_bonus * 0.2)
        
        return sustainability_score
    
    def _calculate_transparency_score(self, fleet_allocation: np.ndarray) -> float:
        """Calculate transparency score"""
        # Base transparency
        transparency = 0.8
        
        # Adjust for complexity
        complexity = np.sum(fleet_allocation)
        if complexity > 10:
            transparency *= 0.9
        
        # Adjust for stakeholder diversity
        stakeholder_count = len(self.constitutional_ai.stakeholders)
        if stakeholder_count > 4:
            transparency *= 0.95
        
        return min(1.0, transparency)
    
    def _calculate_total_ethical_score(self, objectives: Dict[str, float], 
                                        ethical_compliance: Dict[str, float],
                                        equity_score: float, sustainability_score: float,
                                        transparency_score: float) -> float:
        """Calculate total ethical score"""
        # Weighted combination of all factors
        objective_score = np.mean(list(objectives.values()))
        compliance_score = np.mean(list(ethical_compliance.values()))
        
        total_score = (
            objective_score * 0.3 +
            compliance_score * 0.3 +
            equity_score * 0.2 +
            sustainability_score * 0.1 +
            transparency_score * 0.1
        )
        
        return total_score
    
    def _generate_ethical_reasoning(self, fleet_allocation: np.ndarray, 
                                   objectives: Dict[str, float], 
                                   ethical_compliance: Dict[str, float],
                                   equity_score: float) -> List[str]:
        """Generate ethical reasoning for the decision"""
        reasoning = []
        
        # Objective-based reasoning
        best_objective = max(objectives, key=objectives.get)
        reasoning.append(f"Optimized for {best_objective} objective with score {objectives[best_objective]:.3f}")
        
        # Ethical compliance reasoning
        avg_compliance = np.mean(list(ethical_compliance.values()))
        reasoning.append(f"Achieved {avg_compliance:.1%} ethical compliance across all principles")
        
        # Equity reasoning
        reasoning.append(f"Stakeholder equity score of {equity_score:.3f} ensures fair distribution")
        
        # Fleet composition reasoning
        total_vehicles = np.sum(fleet_allocation)
        reasoning.append(f"Selected fleet of {int(total_vehicles)} vehicles balances all objectives")
        
        # Constraint satisfaction reasoning
        constraints_satisfied = all(score >= 0.8 for score in ethical_compliance.values())
        if constraints_satisfied:
            reasoning.append("All ethical constraints satisfied with acceptable margins")
        else:
            reasoning.append("Some ethical constraints require attention and improvement")
        
        return reasoning

# Initialize multi-objective optimizer
optimizer = MultiObjectiveOptimizer(constitutional_ai)
```
