## Mathematical Formulation

### Simple Moving Average (SMA)
\[
F_{t+1} = \frac{1}{n} \sum_{i=0}^{n-1} D_{t-i} = \frac{D_t + D_{t-1} + D_{t-2} + \ldots + D_{t-n+1}}{n}
\]

where:
- $F_{t+1}$ = forecast for period $t+1$
- $D_t$ = actual demand in period $t$
- $n$ = number of periods in the moving average

### Weighted Moving Average (WMA)
\[
F_{t+1} = \sum_{i=0}^{n-1} w_i \cdot D_{t-i}
\]

subject to:
\[
\sum_{i=0}^{n-1} w_i = 1, \quad w_i \geq 0 \quad \forall i
\]

### Error Metrics

**Forecast Error:**
\[
E_t = D_t - F_t
\]

**Mean Absolute Deviation (MAD):**
\[
MAD = \frac{1}{m} \sum_{t=1}^{m} |E_t|
\]

**Mean Squared Error (MSE):**
\[
MSE = \frac{1}{m} \sum_{t=1}^{m} E_t^2
\]

**Mean Absolute Percentage Error (MAPE):**
\[
MAPE = \frac{1}{m} \sum_{t=1}^{m} \left|\frac{E_t}{D_t}\right| \times 100\%
\]

```python
# Define the demand data from the problem example
# Wireless headphone demand for 12 weeks
demand_data = [450, 380, 520, 610, 485, 390, 475, 550, 425, 480, 515, 460]
weeks = list(range(1, 13))  # Week numbers 1-12

print("=== Demand Data Analysis ===")
print(f"Weekly demand data: {demand_data}")
print(f"Number of periods: {len(demand_data)}")
print(f"Mean demand: {np.mean(demand_data):.1f} units")
print(f"Standard deviation: {np.std(demand_data):.1f} units")
print(f"Coefficient of variation: {np.std(demand_data)/np.mean(demand_data):.3f}")
```

```python
class MovingAverageForecaster:
    """
    Mathematical implementation of Simple and Weighted Moving Average forecasting.
    
    This class provides the foundational mathematical methods for demand forecasting
    based on moving averages, following the pen-and-paper approach.
    """
    
    def __init__(self, demand_data: List[float]):
        """
        Initialize the forecaster with historical demand data.
        
        Args:
            demand_data: List of historical demand values
        """
        self.demand_data = demand_data
        self.periods = len(demand_data)
        
    def simple_moving_average(self, window_size: int, forecast_period: int) -> float:
        """
        Calculate Simple Moving Average forecast.
        
        Mathematical formula: F_{t+1} = (1/n) * sum(D_{t-i}) for i=0 to n-1
        
        Args:
            window_size: Number of periods to average (n)
            forecast_period: Period to forecast (t+1)
            
        Returns:
            Forecast value for the specified period
        """
        if forecast_period <= window_size:
            raise ValueError(f"Cannot forecast period {forecast_period} with window size {window_size}")
            
        # Get the most recent 'window_size' periods before forecast_period
        start_idx = forecast_period - window_size - 1
        end_idx = forecast_period - 1
        
        recent_data = self.demand_data[start_idx:end_idx]
        forecast = sum(recent_data) / window_size
        
        return forecast
    
    def weighted_moving_average(self, window_size: int, weights: List[float], 
                                forecast_period: int) -> float:
        """
        Calculate Weighted Moving Average forecast.
        
        Mathematical formula: F_{t+1} = sum(w_i * D_{t-i}) for i=0 to n-1
        Constraint: sum(w_i) = 1, w_i >= 0
        
        Args:
            window_size: Number of periods to include
            weights: List of weights (should sum to 1)
            forecast_period: Period to forecast
            
        Returns:
            Weighted forecast value
        """
        if len(weights) != window_size:
            raise ValueError(f"Number of weights ({len(weights)}) must equal window size ({window_size})")
            
        if abs(sum(weights) - 1.0) > 1e-6:
            raise ValueError(f"Weights must sum to 1.0, current sum: {sum(weights)}")
            
        if forecast_period <= window_size:
            raise ValueError(f"Cannot forecast period {forecast_period} with window size {window_size}")
            
        # Get the most recent periods and apply weights
        start_idx = forecast_period - window_size - 1
        end_idx = forecast_period - 1
        
        recent_data = self.demand_data[start_idx:end_idx]
        
        # Apply weights: most recent data gets first weight
        forecast = sum(w * d for w, d in zip(weights, recent_data))
        
        return forecast
    
    def calculate_forecast_errors(self, forecasts: List[float], 
                                  start_period: int) -> Dict[str, float]:
        """
        Calculate forecast accuracy metrics.
        
        Args:
            forecasts: List of forecast values
            start_period: Starting period for error calculation
            
        Returns:
            Dictionary containing error metrics
        """
        actual_values = self.demand_data[start_period-1:start_period-1+len(forecasts)]
        errors = [actual - forecast for actual, forecast in zip(actual_values, forecasts)]
        
        # Calculate error metrics
        mad = np.mean(np.abs(errors))
        mse = np.mean(np.square(errors))
        rmse = np.sqrt(mse)
        
        # Calculate MAPE (avoiding division by zero)
        mape_values = []
        for actual, error in zip(actual_values, errors):
            if actual != 0:
                mape_values.append(abs(error / actual) * 100)
        mape = np.mean(mape_values) if mape_values else 0
        
        return {
            'MAD': mad,
            'MSE': mse,
            'RMSE': rmse,
            'MAPE': mape,
            'errors': errors
        }

print("MovingAverageForecaster class defined successfully")
```

```python
# Initialize the forecaster with our demand data
forecaster = MovingAverageForecaster(demand_data)

# Example: Calculate 3-period SMA and WMA for Week 13
window_size = 3
forecast_period = 13

# Define weights for WMA (most recent to oldest)
wma_weights = [0.5, 0.3, 0.2]  # Week 12: 0.5, Week 11: 0.3, Week 10: 0.2

print("=== Concrete Mathematical Example ===")
print(f"\n3-Period Simple Moving Average for Week {forecast_period}:")
print(f"Formula: F_{forecast_period} = (D_{forecast_period-1} + D_{forecast_period-2} + D_{forecast_period-3}) / 3")
print(f"F_{forecast_period} = ({demand_data[11]} + {demand_data[10]} + {demand_data[9]}) / 3")
print(f"F_{forecast_period} = ({demand_data[11]} + {demand_data[10]} + {demand_data[9]}) / 3 = {sum(demand_data[9:12])}/3 = {sum(demand_data[9:12])/3:.1f}")

sma_forecast = forecaster.simple_moving_average(window_size, forecast_period)
print(f"\nCalculated SMA forecast: {sma_forecast:.1f} units")

print(f"\n3-Period Weighted Moving Average for Week {forecast_period}:")
print(f"Weights: {wma_weights} (most recent to oldest)")
print(f"Formula: F_{forecast_period} = 0.5*D_{forecast_period-1} + 0.3*D_{forecast_period-2} + 0.2*D_{forecast_period-3}")
print(f"F_{forecast_period} = 0.5*{demand_data[11]} + 0.3*{demand_data[10]} + 0.2*{demand_data[9]}")
wma_manual = 0.5 * demand_data[11] + 0.3 * demand_data[10] + 0.2 * demand_data[9]
print(f"F_{forecast_period} = {0.5 * demand_data[11]} + {0.3 * demand_data[10]} + {0.2 * demand_data[9]} = {wma_manual:.1f}")

wma_forecast = forecaster.weighted_moving_average(window_size, wma_weights, forecast_period)
print(f"\nCalculated WMA forecast: {wma_forecast:.1f} units")

print(f"\nDifference between methods: {sma_forecast - wma_forecast:.1f} units")
```

```python
# Backtest analysis for periods 4-12 (as shown in the mathematical example)
print("=== Historical Backtest Analysis ===")
print("\n3-Period SMA Performance for Periods 4-12:")

# Calculate SMA forecasts for periods 4-12
sma_forecasts = []
for period in range(4, 13):
    forecast = forecaster.simple_moving_average(3, period)
    sma_forecasts.append(forecast)

# Calculate WMA forecasts for periods 4-12
wma_forecasts = []
for period in range(4, 13):
    forecast = forecaster.weighted_moving_average(3, wma_weights, period)
    wma_forecasts.append(forecast)

# Create comparison table
print("\nDetailed Comparison Table:")
print("Period | Actual | SMA Forecast | WMA Forecast | SMA Error | WMA Error")
print("-" * 70)

for i, period in enumerate(range(4, 13)):
    actual = demand_data[period-1]
    sma_f = sma_forecasts[i]
    wma_f = wma_forecasts[i]
    sma_error = actual - sma_f
    wma_error = actual - wma_f
    
    print(f"{period:6d} | {actual:6.0f} | {sma_f:11.1f} | {wma_f:12.1f} | {sma_error:9.1f} | {wma_error:10.1f}")

# Calculate error metrics
sma_errors = forecaster.calculate_forecast_errors(sma_forecasts, 4)
wma_errors = forecaster.calculate_forecast_errors(wma_forecasts, 4)

print(f"\nError Metrics Summary:")
print(f"SMA - MAD: {sma_errors['MAD']:.2f}, RMSE: {sma_errors['RMSE']:.2f}, MAPE: {sma_errors['MAPE']:.2f}%")
print(f"WMA - MAD: {wma_errors['MAD']:.2f}, RMSE: {wma_errors['RMSE']:.2f}, MAPE: {wma_errors['MAPE']:.2f}%")

print(f"\nImprovement with WMA:")
mad_improvement = ((sma_errors['MAD'] - wma_errors['MAD']) / sma_errors['MAD']) * 100
rmse_improvement = ((sma_errors['RMSE'] - wma_errors['RMSE']) / sma_errors['RMSE']) * 100
mape_improvement = ((sma_errors['MAPE'] - wma_errors['MAPE']) / sma_errors['MAPE']) * 100

print(f"MAD improvement: {mad_improvement:.1f}%")
print(f"RMSE improvement: {rmse_improvement:.1f}%")
print(f"MAPE improvement: {mape_improvement:.1f}%")
```

```python
# Sensitivity analysis: Test different window sizes and weight combinations
print("=== Sensitivity Analysis ===")

# Test different window sizes
window_sizes = [2, 3, 4, 5]
results = []

for window in window_sizes:
    # Simple Moving Average
    sma_forecasts_window = []
    for period in range(window+1, 13):
        forecast = forecaster.simple_moving_average(window, period)
        sma_forecasts_window.append(forecast)
    
    sma_errors = forecaster.calculate_forecast_errors(sma_forecasts_window, window+1)
    
    # Weighted Moving Average with exponentially decreasing weights
    weights = []
    for i in range(window):
        # Exponential decay: more recent periods get higher weights
        weight = 2 ** (-(window-1-i))
        weights.append(weight)
    
    # Normalize weights to sum to 1
    total_weight = sum(weights)
    weights = [w/total_weight for w in weights]
    
    wma_forecasts_window = []
    for period in range(window+1, 13):
        forecast = forecaster.weighted_moving_average(window, weights, period)
        wma_forecasts_window.append(forecast)
    
    wma_errors = forecaster.calculate_forecast_errors(wma_forecasts_window, window+1)
    
    results.append({
        'Window_Size': window,
        'SMA_MAD': sma_errors['MAD'],
        'SMA_MAPE': sma_errors['MAPE'],
        'WMA_MAD': wma_errors['MAD'],
        'WMA_MAPE': wma_errors['MAPE'],
        'Weights': weights
    })

# Display sensitivity analysis results
print("\nSensitivity Analysis Results:")
print("Window | SMA MAD | SMA MAPE | WMA MAD | WMA MAPE | Weights (most recent to oldest)")
print("-" * 85)

for result in results:
    weights_str = ", ".join([f"{w:.3f}" for w in result['Weights']])
    print(f"{result['Window_Size']:6d} | {result['SMA_MAD']:7.2f} | {result['SMA_MAPE']:8.2f}% | {result['WMA_MAD']:7.2f} | {result['WMA_MAPE']:8.2f}% | [{weights_str}]")

# Find best performing configuration
best_wma = min(results, key=lambda x: x['WMA_MAD'])
print(f"\nBest WMA configuration: Window size {best_wma['Window_Size']} with MAD {best_wma['WMA_MAD']:.2f}")
```

### Mathematical Insights

The mathematical formulation demonstrates several important principles:

1. **Weight Optimization**: The constraint $\sum w_i = 1$ ensures the forecast scale matches historical demand
2. **Error Minimization**: MAD provides robust error measurement less sensitive to outliers
3. **Trade-off Balance**: Window size selection balances stability vs responsiveness
4. **Temporal Relevance**: Exponentially decreasing weights reflect recency importance
