# 55. The Newsvendor Problem

## Tier 2: Priority-Based Marginal Analysis Heuristic

### Key Assumptions
- Discrete demand scenarios with known probabilities
- Marginal analysis approach for decision making
- Priority-based evaluation of order quantities
- Step-by-step incremental decision process
- Focus on marginal profit vs marginal cost

### Approach
1. Define discrete demand scenarios with probabilities
2. Calculate marginal profit for each additional unit
3. Add units while marginal profit exceeds marginal cost
4. Find optimal stopping point
5. Compare with analytical solution

### Concrete Example
Marina's ski jacket problem:
- Demand: 800 (20%), 1000 (50%), 1200 (30%)
- Same cost structure as Tier 1
- Marginal analysis unit by unit

### Why this Tier exists
- Intuitive step-by-step decision making
- Educational value for marginal analysis
- Handles discrete demand naturally
- Transparent decision logic

```python
import numpy as np
import matplotlib.pyplot as plt
print("Libraries imported")
```

```python
# Problem setup
selling_price = 320.0
unit_cost = 180.0
salvage_value = 90.0

# Demand scenarios
demand_scenarios = [(800, 0.2), (1000, 0.5), (1200, 0.3)]

print("Problem setup complete")
```

```python
# Marginal analysis calculation
def marginal_profit(order_qty, scenarios, sp, uc, sv):
    mp = 0.0
    for demand, prob in scenarios:
        if demand >= order_qty:
            unit_profit = sp - uc
        else:
            unit_profit = sv - uc
        mp += unit_profit * prob
    return mp

# Find optimal quantity
optimal_qty = 0
for qty in range(1, 1500):
    if marginal_profit(qty, demand_scenarios, selling_price, unit_cost, salvage_value) > 0:
        optimal_qty = qty
    else:
        break

print(f"Optimal order quantity: {optimal_qty}")
```

```python
# Visualization
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

# Calculate marginal profits for first 50 units
units = range(1, 51)
marginal_profits = [marginal_profit(q, demand_scenarios, selling_price, unit_cost, salvage_value) for q in units]

ax.bar(units, marginal_profits, alpha=0.7)
ax.axhline(y=0, color='red', linestyle='--', label='Break-even')
ax.axvline(x=optimal_qty, color='green', linestyle='--', label=f'Optimal: {optimal_qty}')
ax.set_xlabel('Unit Number')
ax.set_ylabel('Marginal Profit ($)')
ax.set_title('Marginal Profit Analysis')
ax.legend()
ax.grid(True, alpha=0.3)

plt.show()
print("Visualization complete")
```

## Summary

The priority-based marginal analysis provides an intuitive approach:

- **Optimal Order**: ~1,000 units
- **Logic**: Add units while marginal profit > 0
- **Advantage**: Easy to understand step-by-step process
- **Educational**: Excellent for teaching marginal thinking
- **Comparison**: Close to analytical solution from Tier 1
