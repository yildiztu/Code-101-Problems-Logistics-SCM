# 66. The Workforce Scheduling for Peak/Off-Peak Problem
## Tier 8: Value-Aligned & Ethical Framework (The Fair Scheduler)

### Key assumptions
- Workforce scheduling must balance multiple stakeholder interests
- Ethical constraints and fairness principles guide optimization decisions
- Employee well-being and customer satisfaction are equally important
- Sustainability and social responsibility are integral to business operations

### Approach (step-by-step)
1. **Stakeholder Analysis**: Identify all affected parties and their interests
2. **Ethical Principles**: Define fairness, equity, and sustainability constraints
3. **Multi-Objective Optimization**: Balance competing stakeholder utilities
4. **Constraint Enforcement**: Ensure ethical compliance in all solutions
5. **Fairness Metrics**: Quantify and monitor equity across all dimensions
6. **Continuous Improvement**: Learn from outcomes and refine ethical framework

### What to look for in the results
- Stakeholder utility optimization with ethical constraints
- Fairness metrics and equity analysis across different groups
- Sustainability assessment and environmental impact
- Ethical compliance scoring and constraint satisfaction
- Trade-off analysis between efficiency and fairness

### Concrete example (from the source)
Value-Aligned & Ethical Framework:
- **Employee Equity**: Fair workload distribution and opportunity access
- **Customer Justice**: Equal service quality across all customer segments
- **Community Impact**: Consider local employment and economic contributions
- **Environmental Sustainability**: Minimize carbon footprint and resource waste
- **Shareholder Returns**: Ensure business viability while maintaining ethical standards

```python
from typing import Tuple, List, Dict, Set

# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass
from typing import List, Dict, Tuple, Optional
import warnings
warnings.filterwarnings('ignore')

# Set style for visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
@dataclass
class Stakeholder:
    """Represents a stakeholder in the workforce scheduling system"""
    name: str
    utility_function: callable
    weight: float
    constraints: List[str]
    
@dataclass
class EthicalConstraint:
    """Represents an ethical constraint in the scheduling system"""
    name: str
    constraint_function: callable
    penalty_weight: float
    description: str
    
@dataclass
class ValueMetric:
    """Represents a value metric for measuring ethical performance"""
    name: str
    metric_function: callable
    target_value: float
    tolerance: float
```

```python
class ValueAlignedScheduler:
    """Value-Aligned & Ethical Framework for Workforce Scheduling"""
    
    def __init__(self):
        self.stakeholders = []
        self.ethical_constraints = []
        self.value_metrics = []
        self.schedule_history = []
        
    def add_stakeholder(self, stakeholder: Stakeholder):
        """Add a stakeholder to the system"""
        self.stakeholders.append(stakeholder)
        
    def add_ethical_constraint(self, constraint: EthicalConstraint):
        """Add an ethical constraint to the system"""
        self.ethical_constraints.append(constraint)
        
    def add_value_metric(self, metric: ValueMetric):
        """Add a value metric to the system"""
        self.value_metrics.append(metric)
        
    def calculate_stakeholder_utility(self, schedule: np.ndarray, 
                                    stakeholder: Stakeholder) -> float:
        """Calculate utility for a specific stakeholder"""
        return stakeholder.utility_function(schedule)
        
    def check_ethical_compliance(self, schedule: np.ndarray) -> Tuple[bool, float]:
        """Check if schedule complies with ethical constraints"""
        total_penalty = 0.0
        compliant = True
        
        for constraint in self.ethical_constraints:
            if not constraint.constraint_function(schedule):
                compliant = False
                total_penalty += constraint.penalty_weight
                
        return compliant, total_penalty
        
    def calculate_value_metrics(self, schedule: np.ndarray) -> Dict[str, float]:
        """Calculate all value metrics for a schedule"""
        metrics = {}
        for metric in self.value_metrics:
            metrics[metric.name] = metric.metric_function(schedule)
        return metrics
        
    def optimize_schedule(self, demand_schedule: np.ndarray, 
                         employee_pool: np.ndarray) -> np.ndarray:
        """Optimize schedule considering all stakeholders and ethical constraints"""
        best_schedule = None
        best_score = -np.inf
        
        # Generate multiple schedule candidates
        for _ in range(100):
            # Create candidate schedule
            candidate = self._generate_candidate_schedule(demand_schedule, employee_pool)
            
            # Calculate overall score
            score = self._evaluate_schedule(candidate)
            
            # Update best solution
            if score > best_score:
                best_score = score
                best_schedule = candidate.copy()
                
        return best_schedule
        
    def _generate_candidate_schedule(self, demand: np.ndarray, 
                                   employees: np.ndarray) -> np.ndarray:
        """Generate a candidate schedule"""
        periods = len(demand)
        schedule = np.zeros((len(employees), periods))
        
        for p in range(periods):
            required = int(demand[p])
            available = len(employees)
            
            if required <= available:
                # Randomly assign employees to meet demand
                assigned = np.random.choice(available, required, replace=False)
                schedule[assigned, p] = 1
            else:
                # Assign all available employees
                schedule[:, p] = 1
                
        return schedule
        
    def _evaluate_schedule(self, schedule: np.ndarray) -> float:
        """Evaluate schedule based on stakeholder utilities and ethical compliance"""
        if schedule is None:
            return -np.inf
            
        total_utility = 0.0
        
        # Calculate stakeholder utilities
        for stakeholder in self.stakeholders:
            utility = self.calculate_stakeholder_utility(schedule, stakeholder)
            total_utility += stakeholder.weight * utility
            
        # Check ethical compliance
        compliant, penalty = self.check_ethical_compliance(schedule)
        if not compliant:
            total_utility -= penalty
            
        return total_utility
```

```python
# Define stakeholder utility functions
def employee_utility(schedule: np.ndarray) -> float:
    """Employee utility: prefer balanced workloads and fair distribution"""
    workload_balance = 1.0 - np.std(np.sum(schedule, axis=1)) / np.mean(np.sum(schedule, axis=1))
    return workload_balance

def customer_utility(schedule: np.ndarray) -> float:
    """Customer utility: prefer adequate staffing during all periods"""
    period_coverage = np.mean(np.sum(schedule, axis=0) > 0)
    return period_coverage

def shareholder_utility(schedule: np.ndarray) -> float:
    """Shareholder utility: prefer cost efficiency"""
    total_staff = np.sum(schedule)
    return 1.0 / (1.0 + total_staff)  # Lower staff = higher utility

def community_utility(schedule: np.ndarray) -> float:
    """Community utility: prefer local employment and stability"""
    employment_stability = np.mean(np.sum(schedule, axis=1) > 0)
    return employment_stability
```

```python
# Define ethical constraint functions
def fairness_constraint(schedule: np.ndarray) -> bool:
    """Ensure fair workload distribution among employees"""
    workloads = np.sum(schedule, axis=1)
    max_workload = np.max(workloads)
    min_workload = np.min(workloads[workloads > 0]) if np.any(workloads > 0) else 0
    return max_workload <= 2 * min_workload if min_workload > 0 else True

def sustainability_constraint(schedule: np.ndarray) -> bool:
    """Ensure environmentally sustainable scheduling"""
    # Limit total staffing to reduce environmental impact
    total_staff = np.sum(schedule)
    return total_staff <= 50  # Maximum sustainable staffing level

def equity_constraint(schedule: np.ndarray) -> bool:
    """Ensure equitable access to work opportunities"""
    workloads = np.sum(schedule, axis=1)
    # Ensure at least 80% of employees get some work
    employed_count = np.sum(workloads > 0)
    total_employees = len(workloads)
    return employed_count >= 0.8 * total_employees
```

```python
# Initialize the Value-Aligned Scheduler
scheduler = ValueAlignedScheduler()

# Add stakeholders
scheduler.add_stakeholder(Stakeholder("Employees", employee_utility, 0.3, []))
scheduler.add_stakeholder(Stakeholder("Customers", customer_utility, 0.3, []))
scheduler.add_stakeholder(Stakeholder("Shareholders", shareholder_utility, 0.2, []))
scheduler.add_stakeholder(Stakeholder("Community", community_utility, 0.2, []))

# Add ethical constraints
scheduler.add_ethical_constraint(EthicalConstraint("Fairness", fairness_constraint, 10.0, "Fair workload distribution"))
scheduler.add_ethical_constraint(EthicalConstraint("Sustainability", sustainability_constraint, 5.0, "Environmental sustainability"))
scheduler.add_ethical_constraint(EthicalConstraint("Equity", equity_constraint, 8.0, "Equitable work opportunities"))

# Create sample data
np.random.seed(42)
periods = 24  # 24-hour scheduling
employees = 20

# Demand pattern (higher during peak hours)
demand_schedule = np.array([
    2, 2, 2, 2, 3, 4, 6, 8, 10, 12, 14, 15,
    15, 14, 12, 10, 8, 6, 4, 3, 2, 2, 2, 2
])

employee_pool = np.arange(employees)

print("Value-Aligned Scheduler initialized with:")
print(f"- {len(scheduler.stakeholders)} stakeholders")
print(f"- {len(scheduler.ethical_constraints)} ethical constraints")
print(f"- {employees} employees for {periods} periods")
```

```python
try:
    try:
        # Optimize the schedule
        optimal_schedule = scheduler.optimize_schedule(demand_schedule, employee_pool)
        
        print("=== Value-Aligned Schedule Optimization Results ===")
        print(f"Schedule shape: {optimal_schedule.shape}")
        print(f"Total staff assigned: {np.sum(optimal_schedule)}")
        print(f"Average staff per period: {np.mean(np.sum(optimal_schedule, axis=0)):.2f}")
        
        # Check ethical compliance
        compliant, penalty = scheduler.check_ethical_compliance(optimal_schedule)
        print(f"\nEthical Compliance: {'✓ PASS' if compliant else '✗ FAIL'}")
        if not compliant:
            print(f"Total penalty: {penalty:.2f}")
        
        # Calculate stakeholder utilities
        print("\n=== Stakeholder Utilities ===")
        for stakeholder in scheduler.stakeholders:
            utility = scheduler.calculate_stakeholder_utility(optimal_schedule, stakeholder)
            print(f"{stakeholder.name}: {utility:.3f}")
    except Exception as e:
        print(f'Error in cell: {e}')
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    # Create comprehensive visualization
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    fig.suptitle('Value-Aligned & Ethical Workforce Scheduling', fontsize=16, fontweight='bold')
    
    # 1. Staffing Schedule Heatmap
    ax1 = axes[0, 0]
    sns.heatmap(optimal_schedule, cmap='Blues', ax=ax1, cbar_kws={'label': 'Assigned'})
    ax1.set_title('Staff Assignment Schedule')
    ax1.set_xlabel('Time Period (Hours)')
    ax1.set_ylabel('Employee ID')
    
    # 2. Demand vs Staffing Comparison
    ax2 = axes[0, 1]
    periods_range = range(len(demand_schedule))
    staff_per_period = np.sum(optimal_schedule, axis=0)
    ax2.plot(periods_range, demand_schedule, 'r-', linewidth=2, label='Demand', marker='o')
    ax2.plot(periods_range, staff_per_period, 'b-', linewidth=2, label='Staff Assigned', marker='s')
    ax2.fill_between(periods_range, demand_schedule, staff_per_period, alpha=0.3, color='green', label='Coverage Gap')
    ax2.set_title('Demand vs Staffing Coverage')
    ax2.set_xlabel('Time Period (Hours)')
    ax2.set_ylabel('Number of Staff')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # 3. Stakeholder Utility Comparison
    ax3 = axes[1, 0]
    stakeholder_names = [s.name for s in scheduler.stakeholders]
    utilities = [scheduler.calculate_stakeholder_utility(optimal_schedule, s) for s in scheduler.stakeholders]
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A']
    bars = ax3.bar(stakeholder_names, utilities, color=colors)
    ax3.set_title('Stakeholder Utility Scores')
    ax3.set_ylabel('Utility Score')
    ax3.set_ylim(0, 1)
    for bar, util in zip(bars, utilities):
        height = bar.get_height()
        ax3.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                 f'{util:.3f}', ha='center', va='bottom')
    
    # 4. Workload Distribution and Fairness
    ax4 = axes[1, 1]
    workloads = np.sum(optimal_schedule, axis=1)
    employee_ids = range(len(workloads))
    ax4.bar(employee_ids, workloads, color='skyblue', alpha=0.7)
    ax4.axhline(y=np.mean(workloads), color='red', linestyle='--', label=f'Mean: {np.mean(workloads):.1f}')
    ax4.axhline(y=np.mean(workloads) + np.std(workloads), color='orange', linestyle='--', label=f'+1 Std: {np.mean(workloads) + np.std(workloads):.1f}')
    ax4.axhline(y=np.mean(workloads) - np.std(workloads), color='orange', linestyle='--', label=f'-1 Std: {np.mean(workloads) - np.std(workloads):.1f}')
    ax4.set_title('Workload Distribution (Fairness Analysis)')
    ax4.set_xlabel('Employee ID')
    ax4.set_ylabel('Total Work Hours')
    ax4.legend()
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Print fairness metrics
    print("\n=== Fairness and Equity Metrics ===")
    print(f"Workload standard deviation: {np.std(workloads):.2f}")
    print(f"Workload coefficient of variation: {np.std(workloads)/np.mean(workloads):.3f}")
    print(f"Employees with work: {np.sum(workloads > 0)}/{len(workloads)} ({100*np.sum(workloads > 0)/len(workloads):.1f}%)")
    print(f"Fairness constraint satisfied: {'✓' if fairness_constraint(optimal_schedule) else '✗'}")
    print(f"Sustainability constraint satisfied: {'✓' if sustainability_constraint(optimal_schedule) else '✗'}")
    print(f"Equity constraint satisfied: {'✓' if equity_constraint(optimal_schedule) else '✗'}")
except Exception as e:
    print(f'Error in cell: {e}')
```
