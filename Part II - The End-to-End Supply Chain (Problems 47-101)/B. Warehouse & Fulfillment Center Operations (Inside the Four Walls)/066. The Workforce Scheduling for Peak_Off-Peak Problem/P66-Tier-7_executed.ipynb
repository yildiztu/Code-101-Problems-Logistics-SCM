# 66. The Workforce Scheduling for Peak/Off-Peak Problem
## Tier 7: The Human-AI Symbiotic Partnership (The Centaur Scheduler)

### Key assumptions
- AI generates multiple high-quality schedule options
- Human managers provide contextual wisdom and expertise
- Explainable AI makes reasoning transparent and understandable
- Collaborative refinement improves both AI and human decisions

### Approach (step-by-step)
1. **AI Generation**: Create diverse, high-quality schedule candidates
2. **Explainable AI**: Provide clear reasoning for each recommendation
3. **Human Review**: Managers evaluate and provide feedback
4. **Collaborative Refinement**: AI adjusts based on human input
5. **Learning Loop**: System improves from human-AI interactions
6. **Final Selection**: Choose optimal schedule through partnership

### What to look for in the results
- Multiple schedule options with different optimization objectives
- Clear explanations of AI reasoning and trade-offs
- Human feedback incorporation and schedule refinement
- Learning and improvement over time
- Quantified benefits of human-AI collaboration

### Concrete example (from the source)
Centaur Scheduler partnership:
- **Schedule A**: Minimum cost using part-time staff
- **Schedule B**: Maximum employee satisfaction
- **Schedule C**: Most robust under demand uncertainty
- **Human Input**: Employee constraints, preferences, operational considerations
- **AI Learning**: Patterns from human decisions improve future recommendations

```python
from typing import Tuple, List, Dict, Set

# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from typing import List, Dict, Tuple, Optional
from dataclasses import dataclass, field
from datetime import datetime, timedelta
import random
import time
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('default')
sns.set_palette("husl")
random.seed(42)
np.random.seed(42)
```

```python
# Define data structures for Human-AI Symbiotic System
@dataclass
class Employee:
    """Employee with preferences and constraints"""
    id: int
    name: str
    skills: List[str]
    hourly_rate: float
    max_hours: int
    preferred_shifts: List[str] = field(default_factory=list)
    unavailable_periods: List[Tuple[float, float]] = field(default_factory=list)
    satisfaction_weight: float = 0.8
    experience_level: str = "intermediate"  # junior, intermediate, senior

@dataclass
class HumanManager:
    """Human Manager with expertise and preferences"""
    id: int
    name: str
    department: str
    experience_years: int
    decision_style: str  # analytical, intuitive, collaborative
    priorities: Dict[str, float] = field(default_factory=dict)
    trust_in_ai: float = 0.7
    risk_tolerance: float = 0.5

@dataclass
class ScheduleOption:
    """AI-generated schedule option with explanations"""
    id: str
    name: str
    schedule: Dict[int, List[float]]  # employee_id -> [start, end]
    objective_scores: Dict[str, float]  # cost, satisfaction, robustness, etc.
    explanation: str
    trade_offs: List[str]
    confidence_score: float
    human_feedback: Optional[Dict] = None
    refined_schedule: Optional[Dict[int, List[float]]] = None
```

```python
# Centaur Scheduler - Human-AI Symbiotic System
class CentaurScheduler:
    """Human-AI Symbiotic Workforce Scheduler"""
    
    def __init__(self, employees: List[Employee], demand_pattern: List[float],
                 cost_params: Dict, time_horizon: float = 480.0):
        """
        Initialize the Centaur Scheduler
        
        Args:
            employees: List of employees
            demand_pattern: Demand pattern over time periods
            cost_params: Cost parameters
            time_horizon: Total time horizon in minutes
        """
        self.employees = employees
        self.demand_pattern = demand_pattern
        self.cost_params = cost_params
        self.time_horizon = time_horizon
        
        # AI components
        self.ai_models = {
            'cost_optimizer': self._create_cost_optimizer(),
            'satisfaction_maximizer': self._create_satisfaction_optimizer(),
            'robustness_analyzer': self._create_robustness_analyzer(),
            'explainability_engine': self._create_explainability_engine()
        }
        
        # Learning database
        self.collaboration_history: List[object] = []
        self.preference_patterns: Dict[int, Dict] = {}  # manager_id -> preferences
        self.successful_patterns: List[Dict] = []  # successful schedule patterns
        
    def _create_cost_optimizer(self) -> Dict:
        """Create AI model for cost optimization"""
        return {
            'type': 'optimization',
            'objective': 'minimize_cost',
            'constraints': ['coverage', 'availability'],
            'preferences': ['part_time_flexible', 'efficiency_focused']
        }
    
    def _create_satisfaction_optimizer(self) -> Dict:
        """Create AI model for employee satisfaction maximization"""
        return {
            'type': 'optimization',
            'objective': 'maximize_satisfaction',
            'constraints': ['minimum_coverage'],
            'preferences': ['preferred_shifts', 'work_life_balance']
        }
    
    def _create_robustness_analyzer(self) -> Dict:
        """Create AI model for robustness analysis"""
        return {
            'type': 'analysis',
            'objective': 'maximize_robustness',
            'scenarios': ['demand_surge', 'staff_shortage', 'extended_hours'],
            'preferences': ['redundancy', 'flexibility']
        }
    
    def _create_explainability_engine(self) -> Dict:
        """Create explainability engine for AI reasoning"""
        return {
            'type': 'explanation',
            'methods': ['feature_importance', 'counterfactuals', 'analogies'],
            'clarity_level': 'high',
            'detail_level': 'comprehensive'
        }
    
    def generate_schedule_options(self, manager: HumanManager, 
                                n_options: int = 3) -> List[object]:
        """
        Generate multiple schedule options using different AI approaches
        
        Args:
            manager: Human manager for personalization
            n_options: Number of options to generate
        
        Returns:
            List of schedule options
        """
        options = []
        
        # Option 1: Cost-optimized schedule
        cost_schedule = self._optimize_for_cost(manager)
        cost_option = self._create_schedule_option(
            "cost_optimized", "Minimum Cost Schedule", cost_schedule, manager
        )
        options.append(cost_option)
        
        # Option 2: Satisfaction-maximized schedule
        satisfaction_schedule = self._optimize_for_satisfaction(manager)
        satisfaction_option = self._create_schedule_option(
            "satisfaction_maximized", "Maximum Employee Satisfaction", 
            satisfaction_schedule, manager
        )
        options.append(satisfaction_option)
        
        # Option 3: Robust schedule
        robust_schedule = self._optimize_for_robustness(manager)
        robust_option = self._create_schedule_option(
            "robust", "Most Robust Schedule", robust_schedule, manager
        )
        options.append(robust_option)
        
        return options
    
    def _optimize_for_cost(self, manager: HumanManager) -> Dict[int, List[float]]:
        """Generate cost-optimized schedule"""
        schedule = {}
        
        # Sort employees by hourly rate (cheapest first)
        sorted_employees = sorted(self.employees, key=lambda e: e.hourly_rate)
        
        # Calculate required staffing for each period
        period_demands = np.array(self.demand_pattern)
        max_demand = int(np.max(period_demands))
        
        # Assign cheapest employees first
        for i, employee in enumerate(sorted_employees):
            if i >= max_demand:
                break
            
            # Assign full shift to maximize efficiency
            schedule[employee.id] = [0, self.time_horizon]
        
        return schedule
    
    def _optimize_for_satisfaction(self, manager: HumanManager) -> Dict[int, List[float]]:
        """Generate satisfaction-maximized schedule"""
        schedule = {}
        
        # Sort employees by satisfaction weight and experience
        sorted_employees = sorted(self.employees, 
                                 key=lambda e: (e.satisfaction_weight, e.experience_level), 
                                 reverse=True)
        
        # Calculate required staffing
        period_demands = np.array(self.demand_pattern)
        avg_demand = int(np.mean(period_demands))
        
        # Assign preferred shifts when possible
        for employee in sorted_employees[:avg_demand]:
            # Try to assign preferred shift
            if employee.preferred_shifts:
                preferred_shift = employee.preferred_shifts[0]
                if preferred_shift == "morning":
                    schedule[employee.id] = [0, self.time_horizon // 2]
                elif preferred_shift == "afternoon":
                    schedule[employee.id] = [self.time_horizon // 2, self.time_horizon]
                else:  # full_shift
                    schedule[employee.id] = [0, self.time_horizon]
            else:
                # Assign reasonable shift
                schedule[employee.id] = [60, self.time_horizon - 60]
        
        return schedule
    
    def _optimize_for_robustness(self, manager: HumanManager) -> Dict[int, List[float]]:
        """Generate robust schedule with redundancy"""
        schedule = {}
        
        # Calculate staffing with buffer for uncertainty
        period_demands = np.array(self.demand_pattern)
        max_demand = int(np.max(period_demands))
        buffer_staff = max(1, max_demand // 4)  # 25% buffer
        total_staff = max_demand + buffer_staff
        
        # Create overlapping shifts for redundancy
        shift_duration = self.time_horizon // 2
        
        for i, employee in enumerate(self.employees[:total_staff]):
            if i < len(self.employees):
                # Stagger shifts for coverage overlap
                start_time = (i * shift_duration // 2) % (self.time_horizon - shift_duration)
                end_time = start_time + shift_duration
                schedule[employee.id] = [start_time, end_time]
        
        return schedule
```

### Why this Tier exists vs earlier Tiers
The Human-AI Symbiotic Partnership provides unique capabilities:
- **Human wisdom integration**: Combines AI computational power with human expertise
- **Explainable AI**: Transparent reasoning builds trust and understanding
- **Collaborative refinement**: Humans and AI improve each other's decisions
- **Adaptive learning**: System learns from human preferences and patterns
- **Contextual understanding**: Humans provide real-world context AI lacks

### Pros / Cons vs earlier Tiers
**vs Tier 1 (MDP):**
- ✅ **Human contextual knowledge** vs pure mathematical optimization
- ✅ **Adaptive learning** vs static mathematical model
- ✅ **Explainable decisions** vs black-box optimization
- ✅ **Real-world applicability** vs theoretical optimality
- ❌ **No optimality guarantee** unlike exact MDP solution
- ❌ **Subject to human bias** and preferences

**vs Tier 2 (VND):**
- ✅ **Intelligent collaboration** vs mechanical heuristic search
- ✅ **Learning and adaptation** vs fixed neighborhood rules
- ✅ **Human expertise integration** vs algorithmic approach only
- ✅ **Personalized solutions** vs one-size-fits-all optimization
- ❌ **More complex** implementation and coordination
- ❌ **Requires human involvement** and time investment

**vs Tier 3 (GWO):**
- ✅ **Human-guided search** vs purely algorithmic exploration
- ✅ **Contextual constraints** vs mathematical constraints only
- ✅ **Explainable reasoning** vs swarm intelligence behavior
- ✅ **Domain expertise** vs generic optimization
- ❌ **Slower convergence** due to human interaction
- ❌ **Less systematic** than pure metaheuristic search

**vs Tier 4 (Transformer):**
- ✅ **Human-AI collaboration** vs automated forecasting
- ✅ **Contextual decision making** vs pattern recognition
- ✅ **Real-time adaptation** vs offline training
- ✅ **Explainable outcomes** vs neural network black box
- ❌ **Different focus** - decision making vs prediction
- ❌ **Requires human availability** and engagement

**vs Tier 5 (Digital Twin):**
- ✅ **Human insight integration** vs pure simulation
- ✅ **Collaborative decision making** vs automated optimization
- ✅ **Adaptive learning** vs fixed simulation models
- ✅ **Explainable recommendations** vs simulation outputs
- ❌ **Less comprehensive** than full digital twin simulation
- ❌ **Human-dependent** vs automated system operation

### When to use this Tier
- **Complex decision environments** where human expertise is valuable

- **Organizations with strong managerial expertise** and experience
- **Dynamic environments** where context matters more than pure optimization
- **Training and development** environments for building AI trust
