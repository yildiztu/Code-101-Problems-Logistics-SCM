### QUBO Formulation
The cross-docking door assignment problem can be formulated as a QUBO problem:

#### **Decision Variables**
- **Binary Variables**: xᵢⱼ ∈ {0,1} where i represents vehicle and j represents door
- **Assignment Constraints**: Each vehicle assigned to exactly one door
- **Capacity Constraints**: Door capacity limits respected

#### **Objective Function**
```
Minimize: Σᵢⱼ cᵢⱼ xᵢⱼ + λ₁ Σᵢ (Σⱼ xᵢⱼ - 1)² + λ₂ Σⱼ (Σᵢ xᵢⱼ - capⱼ)²
```

Where:
- cᵢⱼ: Cost of assigning vehicle i to door j
- λ₁, λ₂: Penalty weights for constraint violations
- capⱼ: Capacity of door j

````python
@dataclass
class CrossDockingInstance:
    """Cross-docking problem instance for quantum optimization"""
    num_vehicles: int
    num_doors: int
    cost_matrix: np.ndarray
    door_capacities: List[int]
    vehicle_demands: List[int]
    
@dataclass
class QUBOFormulation:
    """QUBO formulation of cross-docking problem"""
    Q_matrix: np.ndarray  # Quadratic coefficient matrix
    linear_terms: np.ndarray  # Linear coefficient vector
    constant_term: float  # Constant term
    num_variables: int  # Total number of binary variables
    variable_mapping: Dict[Tuple[int, int], int]  # Map (vehicle, door) to variable index
    
@dataclass
class QuantumAnnealingResult:
    """Result of quantum annealing optimization"""
    solution_vector: np.ndarray  # Binary solution vector
    objective_value: float  # Objective function value
    assignment_matrix: np.ndarray  # Vehicle-door assignment matrix
    is_feasible: bool  # Solution feasibility
    computation_time: float  # Time taken for computation
    quantum_metrics: Dict[str, float]  # Quantum-specific metrics

print("Quantum optimization data structures defined!")
````

````python
class QUBOConstructor:
    """Constructs QUBO formulation from cross-docking problem"""
    
    def __init__(self, penalty_weight_assignment: float = 100.0, 
                 penalty_weight_capacity: float = 100.0):
        self.lambda_assignment = penalty_weight_assignment
        self.lambda_capacity = penalty_weight_capacity
        
    def construct_qubo(self, instance: CrossDockingInstance) -> QUBOFormulation:
        """Construct QUBO formulation from cross-docking instance"""
        n_vehicles = instance.num_vehicles
        n_doors = instance.num_doors
        n_variables = n_vehicles * n_doors
        
        # Create variable mapping
        variable_mapping = {}
        for i in range(n_vehicles):
            for j in range(n_doors):
                variable_mapping[(i, j)] = i * n_doors + j
        
        # Initialize QUBO components
        Q_matrix = np.zeros((n_variables, n_variables))
        linear_terms = np.zeros(n_variables)
        constant_term = 0.0
        
        # 1. Add objective function terms (assignment costs)
        for i in range(n_vehicles):
            for j in range(n_doors):
                var_idx = variable_mapping[(i, j)]
                linear_terms[var_idx] += instance.cost_matrix[i, j]
        
        # 2. Add assignment constraints (each vehicle assigned to exactly one door)
        for i in range(n_vehicles):
            # Σⱼ xᵢⱼ = 1
            for j1 in range(n_doors):
                for j2 in range(n_doors):
                    if j1 != j2:
                        idx1 = variable_mapping[(i, j1)]
                        idx2 = variable_mapping[(i, j2)]
                        Q_matrix[idx1, idx2] += self.lambda_assignment
            
            # Linear terms for assignment constraint
            for j in range(n_doors):
                var_idx = variable_mapping[(i, j)]
                linear_terms[var_idx] -= 2 * self.lambda_assignment
            
            # Constant term
            constant_term += self.lambda_assignment
        
        # 3. Add capacity constraints (door capacity limits)
        for j in range(n_doors):
            # Σᵢ xᵢⱼ ≤ capⱼ
            for i1 in range(n_vehicles):
                for i2 in range(n_vehicles):
                    if i1 != i2:
                        idx1 = variable_mapping[(i1, j)]
                        idx2 = variable_mapping[(i2, j)]
                        Q_matrix[idx1, idx2] += self.lambda_capacity
            
            # Linear terms for capacity constraint
            for i in range(n_vehicles):
                var_idx = variable_mapping[(i, j)]
                linear_terms[var_idx] -= 2 * instance.door_capacities[j] * self.lambda_capacity
            
            # Constant term
            constant_term += instance.door_capacities[j]**2 * self.lambda_capacity
        
        # Ensure Q_matrix is symmetric
        Q_matrix = (Q_matrix + Q_matrix.T) / 2
        
        return QUBOFormulation(
            Q_matrix=Q_matrix,
            linear_terms=linear_terms,
            constant_term=constant_term,
            num_variables=n_variables,
            variable_mapping=variable_mapping
        )
    
    def evaluate_qubo(self, qubo: QUBOFormulation, solution_vector: np.ndarray) -> float:
        """Evaluate QUBO objective function for given solution"""
        # Quadratic terms
        quadratic = np.dot(solution_vector, np.dot(qubo.Q_matrix, solution_vector))
        
        # Linear terms
        linear = np.dot(qubo.linear_terms, solution_vector)
        
        # Constant term
        total = quadratic + linear + qubo.constant_term
        
        return total

print("QUBOConstructor class defined!")
````

````python
class QuantumAnnealer:
    """Simulated quantum annealer for cross-docking optimization"""
    
    def __init__(self, temperature_schedule: str = "linear", 
                 tunneling_probability: float = 0.1):
        self.temperature_schedule = temperature_schedule
        self.tunneling_prob = tunneling_probability
        self.quantum_metrics = {}
        
    def anneal(self, qubo: QUBOFormulation, num_steps: int = 1000, 
               initial_solution: Optional[np.ndarray] = None) -> QuantumAnnealingResult:
        """Perform quantum annealing optimization"""
        start_time = time.time()
        
        # Initialize solution
        if initial_solution is None:
            current_solution = np.random.randint(0, 2, qubo.num_variables)
        else:
            current_solution = initial_solution.copy()
        
        # Evaluate initial solution
        current_energy = self._evaluate_energy(qubo, current_solution)
        best_solution = current_solution.copy()
        best_energy = current_energy
        
        # Annealing process
        energy_history = [current_energy]
        tunneling_events = 0
        superposition_states = 0
        entanglement_effects = 0
        
        for step in range(num_steps):
            # Calculate temperature (quantum-inspired cooling)
            temperature = self._calculate_temperature(step, num_steps)
            
            # Generate candidate solution with quantum effects
            candidate_solution = self._generate_candidate(
                current_solution, temperature, qubo.num_variables)
            
            # Evaluate candidate
            candidate_energy = self._evaluate_energy(qubo, candidate_solution)
            delta_energy = candidate_energy - current_energy
            
            # Quantum-inspired acceptance criterion
            accepted = False
            
            if delta_energy < 0:
                # Accept better solution
                current_solution = candidate_solution
                current_energy = candidate_energy
                accepted = True
            else:
                # Quantum tunneling: accept worse solutions probabilistically
                if np.random.random() < self.tunneling_prob:
                    # Quantum tunneling through energy barrier
                    tunneling_prob = np.exp(-delta_energy / (temperature + 1e-10))
                    if np.random.random() < tunneling_prob:
                        current_solution = candidate_solution
                        current_energy = candidate_energy
                        accepted = True
                        tunneling_events += 1
            
            # Superposition simulation: occasionally explore multiple states
            if np.random.random() < 0.05:  # 5% chance
                # Simulate superposition by trying multiple flips
                superposition_solution = self._superposition_exploration(
                    current_solution, qubo, temperature)
                superposition_energy = self._evaluate_energy(qubo, superposition_solution)
                
                if superposition_energy < current_energy:
                    current_solution = superposition_solution
                    current_energy = superposition_energy
                    superposition_states += 1
            
            # Entanglement effects: correlated variable updates
            if np.random.random() < 0.03:  # 3% chance
                entangled_solution = self._entanglement_update(
                    current_solution, qubo)
                entangled_energy = self._evaluate_energy(qubo, entangled_solution)
                
                if entangled_energy < current_energy:
                    current_solution = entangled_solution
                    current_energy = entangled_energy
                    entanglement_effects += 1
            
            # Update best solution
            if current_energy < best_energy:
                best_solution = current_solution.copy()
                best_energy = current_energy
            
            energy_history.append(current_energy)
        
        # Convert solution to assignment matrix
        assignment_matrix = self._solution_to_assignment(qubo, best_solution)
        
        # Check feasibility
        is_feasible = self._check_feasibility(assignment_matrix)
        
        # Calculate quantum metrics
        computation_time = time.time() - start_time
        quantum_metrics = {
            "tunneling_events": tunneling_events,
            "superposition_states": superposition_states,
            "entanglement_effects": entanglement_effects,
            "energy_reduction": (energy_history[0] - best_energy) / energy_history[0] * 100,
            "convergence_rate": self._calculate_convergence_rate(energy_history)
        }
        
        return QuantumAnnealingResult(
            solution_vector=best_solution,
            objective_value=best_energy,
            assignment_matrix=assignment_matrix,
            is_feasible=is_feasible,
            computation_time=computation_time,
            quantum_metrics=quantum_metrics
        )
    
    def _calculate_temperature(self, step: int, total_steps: int) -> float:
        """Calculate temperature for current annealing step"""
        if self.temperature_schedule == "linear":
            return 1.0 - step / total_steps
        elif self.temperature_schedule == "exponential":
            return np.exp(-5 * step / total_steps)
        else:  # polynomial
            return (1 - step / total_steps) ** 2
    
    def _generate_candidate(self, current_solution: np.ndarray, 
                          temperature: float, num_variables: int) -> np.ndarray:
        """Generate candidate solution with quantum effects"""
        candidate = current_solution.copy()
        
        # Number of flips depends on temperature (quantum tunneling effect)
        num_flips = max(1, int(temperature * 5))
        
        for _ in range(num_flips):
            # Random variable flip
            var_idx = np.random.randint(0, num_variables)
            candidate[var_idx] = 1 - candidate[var_idx]
        
        return candidate
    
    def _evaluate_energy(self, qubo: QUBOFormulation, solution: np.ndarray) -> float:
        """Evaluate energy (objective function) for solution"""
        # Quadratic terms
        quadratic = np.dot(solution, np.dot(qubo.Q_matrix, solution))
        
        # Linear terms
        linear = np.dot(qubo.linear_terms, solution)
        
        # Constant term
        total = quadratic + linear + qubo.constant_term
        
        return total
    
    def _superposition_exploration(self, current_solution: np.ndarray, 
                                qubo: QUBOFormulation, temperature: float) -> np.ndarray:
        """Simulate quantum superposition exploration"""
        candidate = current_solution.copy()
        
        # Try multiple simultaneous flips (superposition effect)
        num_flips = np.random.randint(2, 5)
        flip_indices = np.random.choice(qubo.num_variables, num_flips, replace=False)
        
        for idx in flip_indices:
            candidate[idx] = 1 - candidate[idx]
        
        return candidate
    
    def _entanglement_update(self, current_solution: np.ndarray, 
                           qubo: QUBOFormulation) -> np.ndarray:
        """Simulate quantum entanglement effects"""
        candidate = current_solution.copy()
        
        # Find strongly coupled variables (high Q_matrix values)
        strong_couplings = np.where(np.abs(qubo.Q_matrix) > np.percentile(np.abs(qubo.Q_matrix), 90))
        
        if len(strong_couplings[0]) > 0:
            # Select a random strong coupling
            idx = np.random.randint(0, len(strong_couplings[0]))
            i, j = strong_couplings[0][idx], strong_couplings[1][idx]
            
            # Correlated flip (entanglement effect)
            if np.random.random() < 0.7:  # 70% correlation
                candidate[i] = 1 - candidate[i]
                candidate[j] = 1 - candidate[j]
        
        return candidate
    
    def _solution_to_assignment(self, qubo: QUBOFormulation, 
                               solution: np.ndarray) -> np.ndarray:
        """Convert binary solution vector to assignment matrix"""
        n_vehicles = len(qubo.variable_mapping) // len(set(k[1] for k in qubo.variable_mapping.keys()))
        n_doors = len(set(k[1] for k in qubo.variable_mapping.keys()))
        
        assignment = np.zeros((n_vehicles, n_doors))
        
        for (i, j), var_idx in qubo.variable_mapping.items():
            assignment[i, j] = solution[var_idx]
        
        return assignment
    
    def _check_feasibility(self, assignment_matrix: np.ndarray) -> bool:
        """Check if assignment matrix satisfies constraints"""
        # Check assignment constraint (each vehicle assigned to exactly one door)
        row_sums = np.sum(assignment_matrix, axis=1)
        if not np.all(row_sums == 1):
            return False
        
        # Check capacity constraints (would need door capacities for full check)
        # For now, just check basic feasibility
        return True
    
    def _calculate_convergence_rate(self, energy_history: List[float]) -> float:
        """Calculate convergence rate from energy history"""
        if len(energy_history) < 10:
            return 0.0
        
        # Calculate rate of energy reduction in final 10% of iterations
        final_10_percent = max(1, len(energy_history) // 10)
        final_energies = energy_history[-final_10_percent:]
        
        if len(final_energies) < 2:
            return 0.0
        
        # Linear regression on final energies
        x = np.arange(len(final_energies))
        slope = np.polyfit(x, final_energies, 1)[0]
        
        return abs(slope)  # Return absolute convergence rate

print("QuantumAnnealer class defined!")
````

````python
try:
    # Compare with classical optimization methods
    print("🔄 Comparing with Classical Optimization Methods...")
    
    # Method 1: Hungarian Algorithm (optimal for assignment problem)
    print("\n🎯 Hungarian Algorithm (Optimal Assignment):")
    start_time = time.time()
    row_indices, col_indices = linear_sum_assignment(cost_matrix)
    hungarian_time = time.time() - start_time
    
    hungarian_cost = cost_matrix[row_indices, col_indices].sum()
    hungarian_assignment = np.zeros((num_vehicles, num_doors))
    hungarian_assignment[row_indices, col_indices] = 1
    
    print(f"  Optimal cost: {hungarian_cost:.2f}")
    print(f"  Computation time: {hungarian_time:.6f} seconds")
    print(f"  Assignments: {list(zip(row_indices+1, col_indices+1))}")
    
    # Method 2: Greedy Algorithm
    print("\n🏃 Greedy Algorithm:")
    start_time = time.time()
    greedy_assignment = np.zeros((num_vehicles, num_doors))
    used_doors = set()
    greedy_cost = 0
    
    # Sort all possible assignments by cost
    all_assignments = []
    for i in range(num_vehicles):
        for j in range(num_doors):
            all_assignments.append((cost_matrix[i, j], i, j))
    
    all_assignments.sort()  # Sort by cost (ascending)
    
    # Greedy selection
    for cost, i, j in all_assignments:
        if i not in [a[0] for a in used_doors] and j not in used_doors:
            greedy_assignment[i, j] = 1
            greedy_cost += cost
            used_doors.add(j)
            if len(used_doors) >= num_vehicles:
                break
    
    greedy_time = time.time() - start_time
    print(f"  Greedy cost: {greedy_cost:.2f}")
    print(f"  Computation time: {greedy_time:.6f} seconds")
    
    # Method 3: Classical Simulated Annealing
    print("\n🌡️ Classical Simulated Annealing:")
    
    class ClassicalAnnealer:
        def __init__(self):
            pass
        
        def anneal(self, cost_matrix, num_steps=1000):
            n_vehicles, n_doors = cost_matrix.shape
            
            # Initialize random assignment
            current_assignment = np.zeros((n_vehicles, n_doors))
            for i in range(n_vehicles):
                j = np.random.randint(0, n_doors)
                current_assignment[i, j] = 1
            
            current_cost = np.sum(current_assignment * cost_matrix)
            best_assignment = current_assignment.copy()
            best_cost = current_cost
            
            for step in range(num_steps):
                temperature = 1.0 - step / num_steps
                
                # Generate neighbor by swapping one assignment
                neighbor_assignment = current_assignment.copy()
                i = np.random.randint(0, n_vehicles)
                old_j = np.argmax(neighbor_assignment[i])
                new_j = np.random.randint(0, n_doors)
                
                neighbor_assignment[i, old_j] = 0
                neighbor_assignment[i, new_j] = 1
                
                neighbor_cost = np.sum(neighbor_assignment * cost_matrix)
                delta_cost = neighbor_cost - current_cost
                
                if delta_cost < 0 or np.random.random() < np.exp(-delta_cost / (temperature + 1e-10)):
                    current_assignment = neighbor_assignment
                    current_cost = neighbor_cost
                    
                    if current_cost < best_cost:
                        best_assignment = current_assignment.copy()
                        best_cost = current_cost
            
            return best_assignment, best_cost
    
    classical_annealer = ClassicalAnnealer()
    start_time = time.time()
    classical_assignment, classical_cost = classical_annealer.anneal(cost_matrix, 1000)
    classical_time = time.time() - start_time
    
    print(f"  Classical SA cost: {classical_cost:.2f}")
    print(f"  Computation time: {classical_time:.3f} seconds")
    
    # Comparison summary
    print(f"\n📊 Optimization Method Comparison:")
    methods = [
        ("Hungarian Algorithm", hungarian_cost, hungarian_time, "Optimal"),
        ("Quantum Annealing", total_cost, result.computation_time, "Quantum-Enhanced"),
        ("Classical SA", classical_cost, classical_time, "Heuristic"),
        ("Greedy", greedy_cost, greedy_time, "Simple")
    ]
    
    print(f"{'Method':<20} {'Cost':<10} {'Time (s)':<12} {'Type':<15}")
    print("-" * 60)
    for method, cost, time_taken, method_type in methods:
        print(f"{method:<20} {cost:<10.2f} {time_taken:<12.6f} {method_type:<15}")
    
    # Calculate optimality gaps
    optimal_cost = hungarian_cost
    print(f"\n🎯 Optimality Analysis:")
    for method, cost, _, _ in methods:
        if method != "Hungarian Algorithm":
            gap = (cost - optimal_cost) / optimal_cost * 100
            print(f"  {method}: {gap:+.2f}% from optimal")
except Exception as e:
    print(f'Error in cell: {e}')
````
