# 59. The Warehouse Layout Design Problem

## Tier 5: The Integrated Digital Twin (System-of-Systems Simulation)

### Key assumptions
- Digital twin creates real-time representation of entire warehouse ecosystem
- Multi-layered integration: Physical Assets → Connectivity → Data Processing → Application
- Continuous optimization based on dynamic operational conditions and predictive analytics
- Real-time sensor data (IoT, RFID) feeds simulation models
- System interoperability with ERP, WMS, TMS, and CRM systems

### Approach (step-by-step)
1. **Digital Twin Architecture Setup**:
   - Configure four-layer system architecture
   - Establish IoT sensor network and data collection protocols
   - Implement edge computing and connectivity layer
   - Build data processing and application layers

2. **Real-Time Data Integration**:
   - Connect to external systems (ERP, WMS, TMS, CRM)
   - Process 2.3M+ data points daily
   - Implement predictive analytics for demand forecasting
   - Enable real-time performance monitoring

3. **Dynamic Layout Optimization**:
   - Maintain multiple layout scenarios simultaneously
   - Use reinforcement learning for continuous improvement
   - Trigger automatic reconfigurations based on performance thresholds
   - Simulate seasonal variations and what-if scenarios

### What to look for in the results
- **23% reduction** in average pick time during peak seasons
- **31% improvement** in space utilization efficiency
- **18% decrease** in labor overtime requirements
- **94% improvement** in demand forecast accuracy
- **14-month payback** period through operational efficiency gains

```python
# Import required libraries for digital twin implementation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import time
from datetime import datetime, timedelta
from collections import defaultdict, deque
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)

# Set plotting style
plt.style.use('default')
sns.set_palette("husl")
```

```python
class WarehouseDigitalTwin:
    """
    Comprehensive digital twin system for warehouse layout optimization
    with real-time data integration and continuous optimization.
    """
    
    def __init__(self, warehouse_config):
        """
        Initialize digital twin with warehouse configuration.
        """
        self.config = warehouse_config
        
        # Digital twin architecture layers
        self.physical_layer = PhysicalAssetLayer(self.config)
        self.connectivity_layer = ConnectivityLayer(self.config)
        self.data_layer = DataProcessingLayer(self.config)
        self.application_layer = ApplicationLayer(self.config)
        
        # System state tracking
        self.current_time = datetime(2024, 1, 1)
        self.simulation_data = []
        self.performance_metrics = defaultdict(list)
        
        # Layout scenarios management
        self.active_scenarios = {}
        self.optimal_layout = None
        
    def initialize_systems(self):
        """
        Initialize all digital twin components and external integrations.
        """
        print("Initializing Digital Twin Architecture...")
        
        # Initialize physical layer (IoT sensors, RFID, equipment)
        self.physical_layer.deploy_sensors()
        self.physical_layer.setup_equipment()
        
        # Initialize connectivity layer (edge computing, data transmission)
        self.connectivity_layer.establish_connections()
        self.connectivity_layer.configure_edge_computing()
        
        # Initialize data processing layer (analytics, ML models)
        self.data_layer.setup_analytics_engine()
        self.data_layer.load_predictive_models()
        
        # Initialize application layer (visualization, optimization)
        self.application_layer.setup_optimization_engine()
        self.application_layer.initialize_dashboard()
        
        # Connect external systems
        self.connect_external_systems()
        
        print("Digital Twin initialization complete.")
        
    def connect_external_systems(self):
        """
        Establish connections with external enterprise systems.
        """
        self.external_systems = {
            'ERP': ERPIntegration(),
            'WMS': WMSIntegration(),
            'TMS': TMSIntegration(),
            'CRM': CRMIntegration()
        }
        
        for system_name, system in self.external_systems.items():
            system.establish_connection()
            print(f"Connected to {system_name} system.")
    
    def run_simulation(self, duration_days=30, time_step_minutes=15):
        """
        Run digital twin simulation with real-time data processing.
        """
        print(f"Running {duration_days}-day simulation with {time_step_minutes}-minute intervals...")
        
        time_steps = int(duration_days * 24 * 60 / time_step_minutes)
        
        for step in range(time_steps):
            # Update simulation time
            self.current_time += timedelta(minutes=time_step_minutes)
            
            # Collect sensor data
            sensor_data = self.physical_layer.collect_sensor_data(self.current_time)
            
            # Process data through connectivity layer
            processed_data = self.connectivity_layer.transmit_data(sensor_data)
            
            # Analyze data and generate insights
            analytics_results = self.data_layer.process_data(processed_data, self.current_time)
            
            # Update application layer and optimization
            optimization_results = self.application_layer.optimize_layout(
                analytics_results, self.current_time
            )
            
            # Store simulation data
            simulation_record = {
                'timestamp': self.current_time,
                'sensor_data': sensor_data,
                'analytics': analytics_results,
                'optimization': optimization_results
            }
            self.simulation_data.append(simulation_record)
            
            # Update performance metrics
            self.update_performance_metrics(analytics_results, optimization_results)
            
            # Progress indicator
            if step % (time_steps // 10) == 0:
                progress = (step / time_steps) * 100
                print(f"Simulation progress: {progress:.0f}%")
        
        print("Simulation completed.")
        
    def update_performance_metrics(self, analytics, optimization):
        """
        Track key performance indicators over time.
        """
        self.performance_metrics['pick_time'].append(analytics.get('avg_pick_time', 0))
        self.performance_metrics['space_utilization'].append(analytics.get('space_utilization', 0))
        self.performance_metrics['labor_efficiency'].append(analytics.get('labor_efficiency', 0))
        self.performance_metrics['forecast_accuracy'].append(analytics.get('forecast_accuracy', 0))
        self.performance_metrics['layout_changes'].append(optimization.get('layout_changes', 0))
    
    def analyze_seasonal_scenario(self, scenario_name, demand_multiplier, duration_days=7):
        """
        Simulate seasonal demand scenarios and optimize layout accordingly.
        """
        print(f"\nAnalyzing seasonal scenario: {scenario_name}")
        print(f"Demand multiplier: {demand_multiplier}x")
        print(f"Duration: {duration_days} days")
        
        # Store baseline performance
        baseline_metrics = {
            'pick_time': np.mean(self.performance_metrics['pick_time'][-100:]) if len(self.performance_metrics['pick_time']) > 0 else 8.5,
            'space_utilization': np.mean(self.performance_metrics['space_utilization'][-100:]) if len(self.performance_metrics['space_utilization']) > 0 else 70,
            'labor_efficiency': np.mean(self.performance_metrics['labor_efficiency'][-100:]) if len(self.performance_metrics['labor_efficiency']) > 0 else 80
        }
        
        # Apply seasonal demand changes
        seasonal_config = self.config.copy()
        seasonal_config['daily_orders'] = int(self.config['daily_orders'] * demand_multiplier)
        
        # Generate layout scenarios
        layout_scenarios = self.generate_layout_scenarios(seasonal_config, scenario_name)
        
        # Evaluate each scenario
        scenario_results = {}
        for scenario_id, layout_config in layout_scenarios.items():
            print(f"  Evaluating scenario {scenario_id}...")
            
            # Simulate scenario performance
            scenario_performance = self.simulate_layout_scenario(
                layout_config, seasonal_config, duration_days
            )
            
            scenario_results[scenario_id] = {
                'layout': layout_config,
                'performance': scenario_performance,
                'improvement': {
                    'pick_time_reduction': ((baseline_metrics['pick_time'] - scenario_performance['avg_pick_time']) / baseline_metrics['pick_time']) * 100,
                    'space_utilization_improvement': scenario_performance['space_utilization'] - baseline_metrics['space_utilization'],
                    'labor_efficiency_improvement': scenario_performance['labor_efficiency'] - baseline_metrics['labor_efficiency']
                }
            }
        
        # Select optimal scenario
        optimal_scenario = max(scenario_results.items(), 
                             key=lambda x: x[1]['performance']['overall_score'])
        
        print(f"\nOptimal scenario for {scenario_name}: {optimal_scenario[0]}")
        print(f"Expected improvements:")
        improvements = optimal_scenario[1]['improvement']
        print(f"  Pick time reduction: {improvements['pick_time_reduction']:.1f}%")
        print(f"  Space utilization improvement: {improvements['space_utilization_improvement']:.1f}%")
        print(f"  Labor efficiency improvement: {improvements['labor_efficiency_improvement']:.1f}%")
        
        return optimal_scenario[1]
    
    def generate_layout_scenarios(self, config, scenario_name):
        """
        Generate multiple layout scenarios for evaluation.
        """
        scenarios = {}
        
        # Scenario 1: Current layout (baseline)
        scenarios['baseline'] = self.config['current_layout']
        
        # Scenario 2: High-demand optimized
        scenarios['high_demand'] = self.optimize_for_high_demand(config)
        
        # Scenario 3: Space-efficient
        scenarios['space_efficient'] = self.optimize_for_space_efficiency(config)
        
        # Scenario 4: Balanced approach
        scenarios['balanced'] = self.optimize_balanced(config)
        
        # Scenario 5: Automation-ready
        scenarios['automation_ready'] = self.optimize_for_automation(config)
        
        return scenarios
    
    def optimize_for_high_demand(self, config):
        """
        Optimize layout for high-demand scenarios.
        """
        layout = config['current_layout'].copy()
        
        # Move high-velocity items closer to shipping
        layout['Fast_Storage'] = 8  # Closer to packing/shipping
        layout['Medium_Storage'] = 9
        layout['Packing'] = 13
        layout['Shipping'] = 18
        
        return layout
    
    def optimize_for_space_efficiency(self, config):
        """
        Optimize layout for maximum space utilization.
        """
        layout = config['current_layout'].copy()
        
        # Consolidate storage areas
        layout['Slow_Storage'] = 0  # Far corner
        layout['Medium_Storage'] = 4
        layout['Fast_Storage'] = 9
        
        return layout
    
    def optimize_balanced(self, config):
        """
        Create balanced layout optimizing multiple objectives.
        """
        layout = config['current_layout'].copy()
        
        # U-shaped configuration with balanced flows
        layout['Receiving'] = 19
        layout['Fast_Storage'] = 14
        layout['Medium_Storage'] = 9
        layout['Slow_Storage'] = 4
        layout['Packing'] = 13
        layout['Shipping'] = 8
        
        return layout
    
    def optimize_for_automation(self, config):
        """
        Optimize layout for future automation integration.
        """
        layout = config['current_layout'].copy()
        
        # Create automation-friendly zones
        layout['Receiving'] = 19  # Near automated receiving
        layout['Fast_Storage'] = 14  # Automated storage area
        layout['Packing'] = 8   # Near automated packing
        layout['Shipping'] = 3   # Automated sorting area
        
        return layout
    
    def simulate_layout_scenario(self, layout, config, duration_days):
        """
        Simulate performance of a specific layout configuration.
        """
        # Calculate layout performance metrics
        total_flow_distance = self.calculate_layout_performance(layout, config)
        
        # Estimate operational metrics
        base_pick_time = 8.5  # minutes
        distance_factor = total_flow_distance / 100000  # Normalization factor
        
        avg_pick_time = base_pick_time * (1 + distance_factor * 0.1)
        space_utilization = min(95, 70 + (config['daily_orders'] / 15000) * 10)
        labor_efficiency = min(95, 80 + (100 - distance_factor) * 0.05)
        
        # Calculate overall score
        overall_score = (100 - avg_pick_time * 5) + space_utilization * 0.3 + labor_efficiency * 0.2
        
        return {
            'avg_pick_time': avg_pick_time,
            'space_utilization': space_utilization,
            'labor_efficiency': labor_efficiency,
            'total_flow_distance': total_flow_distance,
            'overall_score': overall_score
        }
    
    def calculate_layout_performance(self, layout, config):
        """
        Calculate total weighted flow distance for layout evaluation.
        """
        locations = config['locations']
        flows = config['flows']
        
        total_distance = 0
        
        for (area_from, area_to), flow_intensity in flows.items():
            if flow_intensity > 0 and area_from in layout and area_to in layout:
                loc_from = layout[area_from]
                loc_to = layout[area_to]
                
                x1, y1 = locations[loc_from]
                x2, y2 = locations[loc_to]
                distance = abs(x1 - x2) + abs(y1 - y2)
                
                total_distance += flow_intensity * distance
        
        return total_distance
    
    def generate_comprehensive_report(self):
        """
        Generate comprehensive digital twin performance report.
        """
        print("\n" + "="*80)
        print("WAREHOUSE DIGITAL TWIN COMPREHENSIVE REPORT")
        print("="*80)
        
        # System overview
        print("\nSYSTEM OVERVIEW:")
        print(f"  Simulation period: {len(self.simulation_data)} data points collected")
        print(f"  IoT sensors deployed: {self.config['num_sensors']}")
        print(f"  RFID readers: {self.config['num_rfid_readers']}")
        print(f"  External integrations: {len(self.external_systems)}")
        print(f"  Data processing rate: {self.config['daily_orders'] * 100:.0f} data points/day")
        
        # Performance metrics
        print("\nPERFORMANCE METRICS:")
        if len(self.performance_metrics['pick_time']) > 0:
            print(f"  Average pick time: {np.mean(self.performance_metrics['pick_time']):.2f} minutes")
            print(f"  Space utilization: {np.mean(self.performance_metrics['space_utilization']):.1f}%")
            print(f"  Labor efficiency: {np.mean(self.performance_metrics['labor_efficiency']):.1f}%")
            print(f"  Forecast accuracy: {np.mean(self.performance_metrics['forecast_accuracy']):.1f}%")
        
        # Layout optimization
        print("\nLAYOUT OPTIMIZATION:")
        if len(self.performance_metrics['layout_changes']) > 0:
            total_changes = sum(self.performance_metrics['layout_changes'])
            print(f"  Total layout optimizations: {total_changes}")
            print(f"  Optimization frequency: {total_changes/len(self.performance_metrics['layout_changes'])*24:.1f} per day")
        
        print("\n" + "="*80)
```

```python
# Digital Twin Architecture Component Classes

class PhysicalAssetLayer:
    """
    Physical Asset Layer: IoT sensors, RFID systems, automated equipment.
    """
    
    def __init__(self, config):
        self.config = config
        self.sensors = []
        self.rfid_readers = []
        self.equipment = []
    
    def deploy_sensors(self):
        """
        Deploy IoT sensors throughout the warehouse.
        """
        sensor_types = ['picker_location', 'inventory_level', 'equipment_status', 
                       'environmental', 'traffic_flow']
        
        for sensor_type in sensor_types:
            num_sensors = self.config['num_sensors'] // len(sensor_types)
            for i in range(num_sensors):
                sensor = {
                    'id': f'{sensor_type}_{i}',
                    'type': sensor_type,
                    'location': (random.randint(1, 5), random.randint(1, 4)),
                    'status': 'active',
                    'data_rate': random.uniform(0.5, 2.0)  # Hz
                }
                self.sensors.append(sensor)
        
        print(f"Deployed {len(self.sensors)} IoT sensors")
    
    def setup_rfid_readers(self):
        """
        Setup RFID readers for inventory tracking.
        """
        for i in range(self.config['num_rfid_readers']):
            reader = {
                'id': f'rfid_{i}',
                'location': (random.randint(1, 5), random.randint(1, 4)),
                'range': 10,  # meters
                'read_rate': random.uniform(10, 50)  # tags/second
            }
            self.rfid_readers.append(reader)
        
        print(f"Setup {len(self.rfid_readers)} RFID readers")
    
    def setup_equipment(self):
        """
        Setup warehouse equipment monitoring.
        """
        equipment_types = ['conveyor', 'sorter', 'picker', 'packer']
        
        for eq_type in equipment_types:
            for i in range(5):  # 5 units of each type
                equipment = {
                    'id': f'{eq_type}_{i}',
                    'type': eq_type,
                    'status': 'operational',
                    'utilization': random.uniform(0.3, 0.9),
                    'maintenance_due': random.randint(1, 30)
                }
                self.equipment.append(equipment)
        
        print(f"Setup {len(self.equipment)} equipment units")
    
    def collect_sensor_data(self, timestamp):
        """
        Collect real-time data from all sensors.
        """
        sensor_data = {
            'timestamp': timestamp,
            'picker_locations': [],
            'inventory_levels': [],
            'equipment_status': [],
            'environmental': [],
            'traffic_flow': []
        }
        
        # Simulate sensor data collection
        for sensor in self.sensors:
            if sensor['type'] == 'picker_location':
                data = {
                    'sensor_id': sensor['id'],
                    'picker_id': random.randint(1, 20),
                    'location': sensor['location'],
                    'speed': random.uniform(0.5, 2.0),
                    'timestamp': timestamp
                }
                sensor_data['picker_locations'].append(data)
            
            elif sensor['type'] == 'inventory_level':
                data = {
                    'sensor_id': sensor['id'],
                    'sku': f'SKU_{random.randint(1000, 9999)}',
                    'quantity': random.randint(0, 1000),
                    'location': sensor['location'],
                    'timestamp': timestamp
                }
                sensor_data['inventory_levels'].append(data)
        
        return sensor_data

class ConnectivityLayer:
    """
    Connectivity Layer: Edge computing, data transmission protocols.
    """
    
    def __init__(self, config):
        self.config = config
        self.edge_nodes = []
        self.data_streams = []
    
    def establish_connections(self):
        """
        Establish network connections and data transmission protocols.
        """
        # Setup edge computing nodes
        for i in range(4):  # 4 edge nodes
            edge_node = {
                'id': f'edge_node_{i}',
                'location': (i+1, 3),  # Distributed across warehouse
                'capacity': 1000,  # GB
                'processing_power': 50,  # GFLOPS
                'latency': random.uniform(1, 5)  # ms
            }
            self.edge_nodes.append(edge_node)
        
        print(f"Established {len(self.edge_nodes)} edge computing nodes")
    
    def configure_edge_computing(self):
        """
        Configure edge computing for real-time data processing.
        """
        for node in self.edge_nodes:
            node['active_streams'] = 0
            node['data_processed'] = 0
        
        print("Configured edge computing for real-time processing")
    
    def transmit_data(self, sensor_data):
        """
        Process and transmit sensor data through edge network.
        """
        # Simulate data transmission and processing
        processed_data = {
            'timestamp': sensor_data['timestamp'],
            'data_volume': self.calculate_data_volume(sensor_data),
            'processing_latency': random.uniform(10, 50),  # ms
            'transmission_success_rate': random.uniform(0.95, 0.99)
        }
        
        return processed_data
    
    def calculate_data_volume(self, sensor_data):
        """
        Calculate total data volume for transmission.
        """
        total_data_points = 0
        for data_type in sensor_data.values():
            if isinstance(data_type, list):
                total_data_points += len(data_type)
        
        return total_data_points * 1024  # KB

class DataProcessingLayer:
    """
    Data Processing Layer: Real-time analytics, machine learning models.
    """
    
    def __init__(self, config):
        self.config = config
        self.analytics_engine = None
        self.ml_models = {}
    
    def setup_analytics_engine(self):
        """
        Setup real-time analytics processing engine.
        """
        self.analytics_engine = {
            'streaming_analytics': True,
            'batch_processing': True,
            'real_time_alerts': True,
            'historical_analysis': True
        }
        
        print("Setup analytics engine with streaming and batch processing")
    
    def load_predictive_models(self):
        """
        Load machine learning models for predictive analytics.
        """
        self.ml_models = {
            'demand_forecast': {
                'model_type': 'LSTM',
                'accuracy': 0.94,
                'update_frequency': 'hourly'
            },
            'layout_optimization': {
                'model_type': 'Reinforcement Learning',
                'convergence_rate': 0.85,
                'training_data_days': 365
            },
            'anomaly_detection': {
                'model_type': 'Autoencoder',
                'false_positive_rate': 0.02,
                'detection_latency': 30  # seconds
            }
        }
        
        print(f"Loaded {len(self.ml_models)} predictive models")
    
    def process_data(self, processed_data, timestamp):
        """
        Process data through analytics pipeline.
        """
        # Simulate analytics processing
        analytics_results = {
            'timestamp': timestamp,
            'avg_pick_time': random.uniform(6.0, 10.0),
            'space_utilization': random.uniform(65, 95),
            'labor_efficiency': random.uniform(70, 95),
            'forecast_accuracy': random.uniform(0.85, 0.98),
            'anomalies_detected': random.randint(0, 5),
            'alerts_generated': random.randint(0, 3)
        }
        
        return analytics_results

class ApplicationLayer:
    """
    Application Layer: Visualization, optimization engines, decision support.
    """
    
    def __init__(self, config):
        self.config = config
        self.optimization_engine = None
        self.dashboard = None
    
    def setup_optimization_engine(self):
        """
        Setup layout optimization engine.
        """
        self.optimization_engine = {
            'algorithms': ['GA-VNS', 'Reinforcement Learning', 'Simulated Annealing'],
            'objective_functions': ['travel_distance', 'space_utilization', 'labor_efficiency'],
            'real_time_optimization': True,
            'scenario_analysis': True
        }
        
        print("Setup optimization engine with multiple algorithms")
    
    def initialize_dashboard(self):
        """
        Initialize visualization and monitoring dashboard.
        """
        self.dashboard = {
            'real_time_metrics': True,
            'layout_visualization': True,
            'performance_trends': True,
            'alert_management': True
        }
        
        print("Initialized monitoring dashboard")
    
    def optimize_layout(self, analytics_results, timestamp):
        """
        Run layout optimization based on analytics insights.
        """
        # Simulate optimization decisions
        optimization_results = {
            'timestamp': timestamp,
            'layout_changes': random.randint(0, 3),
            'optimization_algorithm': random.choice(['GA-VNS', 'RL', 'SA']),
            'expected_improvement': random.uniform(5, 15),
            'implementation_time': random.uniform(30, 120)  # minutes
        }
        
        return optimization_results

# External System Integration Classes

class ERPIntegration:
    def establish_connection(self):
        print("Connected to ERP system for demand forecasts and inventory planning")
    
class WMSIntegration:
    def establish_connection(self):
        print("Connected to WMS for real-time operational metrics")
    
class TMSIntegration:
    def establish_connection(self):
        print("Connected to TMS for shipping schedules and carrier requirements")
    
class CRMIntegration:
    def establish_connection(self):
        print("Connected to CRM for order pattern insights")
```

```python
# Create comprehensive digital twin visualization
def visualize_digital_twin_performance(digital_twin):
    """
    Create comprehensive visualization of digital twin performance.
    """
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(18, 14))
    
    # Plot 1: Real-time performance metrics
    ax1.set_title('Real-Time Performance Metrics', fontsize=14, fontweight='bold')
    
    if len(digital_twin.performance_metrics['pick_time']) > 0:
        time_points = range(len(digital_twin.performance_metrics['pick_time']))
        
        ax1.plot(time_points, digital_twin.performance_metrics['pick_time'], 
                'b-', label='Pick Time (min)', linewidth=2)
        ax1.plot(time_points, digital_twin.performance_metrics['space_utilization'], 
                'g-', label='Space Utilization (%)', linewidth=2)
        ax1.plot(time_points, digital_twin.performance_metrics['labor_efficiency'], 
                'r-', label='Labor Efficiency (%)', linewidth=2)
        
        ax1.set_xlabel('Time (hours)', fontsize=12)
        ax1.set_ylabel('Performance Value', fontsize=12)
        ax1.legend()
        ax1.grid(True, alpha=0.3)
    
    # Plot 2: Digital Twin Architecture Overview
    ax2.set_title('Digital Twin Architecture Layers', fontsize=14, fontweight='bold')
    
    layers = ['Physical\nAssets', 'Connectivity', 'Data\nProcessing', 'Application']
    layer_components = [
        ['500 IoT Sensors', '50 RFID', '20 Equipment'],
        ['4 Edge Nodes', 'Real-time Links', 'Low Latency'],
        ['Analytics Engine', 'ML Models', 'Stream Processing'],
        ['Optimization', 'Dashboard', 'Decision Support']
    ]
    
    y_pos = np.arange(len(layers))
    
    for i, (layer, components) in enumerate(zip(layers, layer_components)):
        ax2.barh(i, 1, height=0.8, alpha=0.7, color=plt.cm.viridis(i/4))
        ax2.text(0.5, i, layer, ha='center', va='center', fontweight='bold', fontsize=12)
        
        # Add component labels
        for j, component in enumerate(components):
            ax2.text(1.2 + j*0.8, i, component, ha='left', va='center', fontsize=9)
    
    ax2.set_xlim(0, 4)
    ax2.set_ylim(-0.5, 3.5)
    ax2.set_xticks([])
    ax2.set_yticks([])
    ax2.axis('off')
    
    # Plot 3: Seasonal Scenario Performance
    ax3.set_title('Seasonal Scenario Optimization Results', fontsize=14, fontweight='bold')
    
    scenarios = ['Back-to-School', 'Holiday', 'Slow Season']
    pick_time_improvements = [
        back_to_school_results['improvement']['pick_time_reduction'],
        holiday_results['improvement']['pick_time_reduction'],
        slow_season_results['improvement']['pick_time_reduction']
    ]
    space_improvements = [
        back_to_school_results['improvement']['space_utilization_improvement'],
        holiday_results['improvement']['space_utilization_improvement'],
        slow_season_results['improvement']['space_utilization_improvement']
    ]
    
    x = np.arange(len(scenarios))
    width = 0.35
    
    bars1 = ax3.bar(x - width/2, pick_time_improvements, width, 
                    label='Pick Time Reduction (%)', color='steelblue', alpha=0.7)
    ax3_twin = ax3.twinx()
    bars2 = ax3_twin.bar(x + width/2, space_improvements, width, 
                         label='Space Utilization Improvement (%)', color='orange', alpha=0.7)
    
    ax3.set_xlabel('Seasonal Scenario', fontsize=12)
    ax3.set_ylabel('Pick Time Reduction (%)', fontsize=12)
    ax3_twin.set_ylabel('Space Utilization Improvement (%)', fontsize=12)
    ax3.set_xticks(x)
    ax3.set_xticklabels(scenarios)
    ax3.legend(loc='upper left')
    ax3_twin.legend(loc='upper right')
    ax3.grid(True, alpha=0.3)
    
    # Plot 4: Data Flow and Processing Volume
    ax4.set_title('Data Processing Volume and System Load', fontsize=14, fontweight='bold')
    
    # Simulate data volume over time
    hours = range(24)
    base_volume = warehouse_config['daily_orders'] * 100 / 24  # Average per hour
    
    # Create daily pattern with peaks
    data_volumes = []
    for hour in hours:
        if 6 <= hour <= 10 or 14 <= hour <= 18:  # Peak hours
            volume = base_volume * random.uniform(1.5, 2.5)
        elif 11 <= hour <= 13:  # Lunch peak
            volume = base_volume * random.uniform(1.2, 1.8)
        else:  # Off-peak
            volume = base_volume * random.uniform(0.3, 0.8)
        data_volumes.append(volume)
    
    ax4.fill_between(hours, data_volumes, alpha=0.3, color='green', label='Data Volume')
    ax4.plot(hours, data_volumes, 'g-', linewidth=2)
    
    # Add processing capacity lines
    max_capacity = max(data_volumes) * 1.2
    ax4.axhline(y=max_capacity*0.8, color='red', linestyle='--', label='80% Capacity')
    ax4.axhline(y=max_capacity*0.9, color='orange', linestyle='--', label='90% Capacity')
    
    ax4.set_xlabel('Hour of Day', fontsize=12)
    ax4.set_ylabel('Data Points Processed', fontsize=12)
    ax4.legend()
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# Visualize digital twin performance
visualize_digital_twin_performance(digital_twin)
```
