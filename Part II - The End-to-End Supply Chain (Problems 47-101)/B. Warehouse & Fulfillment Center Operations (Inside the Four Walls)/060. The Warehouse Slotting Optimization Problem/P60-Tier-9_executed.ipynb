```python
@dataclass
class QuantumSlottingOptimizer:
    """Quantum optimization system for warehouse slotting"""
    
    def __init__(self, num_skus=200, num_locations=50):
        print(f"Initializing Quantum Slotting Optimizer: {num_skus} SKUs, {num_locations} locations")
        
        # Core components
        self.skus = self._create_skus(num_skus)
        self.locations = self._create_locations(num_locations)
        
        # Quantum parameters
        self.qubo_matrix = None
        self.hamiltonian = None
        self.quantum_params = {
            'annealing_time': 100,  # microseconds
            'temperature_schedule': 'geometric',
            'num_reads': 1000,
            'chain_strength': 2.0
        }
        
        # Optimization results
        self.quantum_solution = None
        self.classical_solution = None
        self.hybrid_solution = None
        
        # Performance metrics
        self.quantum_metrics = {}
        self.classical_metrics = {}
        
        print(f"Quantum optimizer initialized with {len(self.skus)} SKUs and {len(self.locations)} locations")
    
    def _create_skus(self, num_skus: int) -> List[Dict]:
        """Create SKUs with quantum-relevant attributes"""
        
        categories = ['Electronics', 'Clothing', 'Food', 'Books', 'Toys', 'Sports', 'Home', 'Beauty']
        skus = []
        
        for i in range(num_skus):
            category = np.random.choice(categories)
            
            # Realistic velocity distribution
            if category == 'Food':
                velocity = np.random.lognormal(4.5, 0.8)
            elif category == 'Electronics':
                velocity = np.random.lognormal(3.5, 1.2)
            else:
                velocity = np.random.lognormal(3.0, 1.0)
            
            sku = {
                'id': i+1,
                'velocity': max(1, velocity),
                'space_req': np.random.uniform(0.1, 4.0),
                'weight': np.random.uniform(0.1, 50),
                'category': category,
                'price': np.random.uniform(5, 500),
                'quantum_id': i,  # Binary variable index
                'compatibility_group': np.random.randint(0, 5)  # For constraint handling
            }
            
            skus.append(sku)
        
        return skus
    
    def _create_locations(self, num_locations: int) -> List[Dict]:
        """Create locations with quantum-compatible attributes"""
        
        zones = ['Fast-Pick', 'Reserve', 'Cold-Storage', 'High-Security', 'Bulk']
        locations = []
        
        for i in range(num_locations):
            zone = np.random.choice(zones, p=[0.3, 0.4, 0.1, 0.1, 0.1])
            
            if zone == 'Fast-Pick':
                capacity = np.random.uniform(8, 20)
                accessibility = np.random.uniform(1, 3)
            elif zone == 'Reserve':
                capacity = np.random.uniform(15, 40)
                accessibility = np.random.uniform(4, 8)
            elif zone == 'Cold-Storage':
                capacity = np.random.uniform(10, 25)
                accessibility = np.random.uniform(3, 6)
            else:
                capacity = np.random.uniform(20, 50)
                accessibility = np.random.uniform(5, 10)
            
            location = {
                'id': i+1,
                'capacity': capacity,
                'weight_limit': capacity * 25,
                'accessibility': accessibility,
                'zone': zone,
                'quantum_id': i,  # Binary variable index
                'zone_preference': {'Fast-Pick': 1.0, 'Reserve': 0.6, 'Cold-Storage': 0.8, 'High-Security': 0.4, 'Bulk': 0.3}[zone]
            }
            
            locations.append(location)
        
        return locations
    
    def formulate_qubo(self) -> np.ndarray:
        """Formulate the slotting problem as a QUBO matrix"""
        
        print("Formulating QUBO matrix for quantum optimization...")
        
        num_skus = len(self.skus)
        num_locations = len(self.locations)
        
        # Each SKU-location assignment is a binary variable
        # Total variables: num_skus * num_locations
        total_vars = num_skus * num_locations
        
        # Initialize QUBO matrix
        qubo = np.zeros((total_vars, total_vars))
        
        # Objective: Maximize velocity/accessibility while minimizing space utilization
        for i, sku in enumerate(self.skus):
            for j, location in enumerate(self.locations):
                var_idx = i * num_locations + j
                
                # Linear term: negative for maximization (quantum annealers minimize)
                velocity_score = -sku['velocity'] / location['accessibility']
                zone_bonus = -location['zone_preference'] * 10
                
                qubo[var_idx, var_idx] = velocity_score + zone_bonus
        
        # Constraint 1: Each SKU must be assigned to exactly one location
        penalty_weight = 1000  # Large penalty for constraint violation
        
        for i in range(num_skus):
            # Get all location variables for this SKU
            sku_vars = [i * num_locations + j for j in range(num_locations)]
            
            # Add penalty: (sum(x_ij) - 1)^2 = sum(x_ij^2) + 2*sum(x_ij*x_ik) - 2*sum(x_ij) + 1
            # Since x_ij^2 = x_ij for binary variables:
            for j in sku_vars:
                qubo[j, j] += penalty_weight  # Linear term
                qubo[j, j] -= 2 * penalty_weight  # Constant term
            
            # Quadratic terms
            for j1 in sku_vars:
                for j2 in sku_vars:
                    if j1 != j2:
                        qubo[j1, j2] += 2 * penalty_weight
        
        # Constraint 2: Location capacity constraints
        for j, location in enumerate(self.locations):
            # Get all SKU variables for this location
            loc_vars = [i * num_locations + j for i in range(num_skus)]
            
            # Add penalty for exceeding capacity
            for i1_idx, i1 in enumerate(loc_vars):
                sku1 = self.skus[i1_idx // num_locations]
                
                # Linear term for capacity
                if sku1['space_req'] > location['capacity']:
                    qubo[i1, i1] += penalty_weight * 10  # Very large penalty
                
                # Quadratic terms for pairwise capacity
                for i2_idx, i2 in enumerate(loc_vars):
                    if i2_idx > i1_idx:  # Avoid double counting
                        sku2 = self.skus[i2_idx // num_locations]
                        
                        if sku1['space_req'] + sku2['space_req'] > location['capacity']:
                            qubo[i1, i2] += penalty_weight * 5
                            qubo[i2, i1] += penalty_weight * 5
        
        # Constraint 3: Weight constraints
        for j, location in enumerate(self.locations):
            loc_vars = [i * num_locations + j for i in range(num_skus)]
            
            for i1_idx, i1 in enumerate(loc_vars):
                sku1 = self.skus[i1_idx // num_locations]
                
                if sku1['weight'] > location['weight_limit']:
                    qubo[i1, i1] += penalty_weight * 10
                
                for i2_idx, i2 in enumerate(loc_vars):
                    if i2_idx > i1_idx:
                        sku2 = self.skus[i2_idx // num_locations]
                        
                        if sku1['weight'] + sku2['weight'] > location['weight_limit']:
                            qubo[i1, i2] += penalty_weight * 5
                            qubo[i2, i1] += penalty_weight * 5
        
        self.qubo_matrix = qubo
        print(f"QUBO matrix created: {total_vars} variables, {np.count_nonzero(qubo)} non-zero elements")
        
        return qubo
    
    def quantum_annealing_simulation(self) -> Dict[str, Any]:
        """Simulate quantum annealing process"""
        
        print("\n=== QUANTUM ANNEALING SIMULATION ===")
        
        if self.qubo_matrix is None:
            self.formulate_qubo()
        
        start_time = time.time()
        
        # Simulated quantum annealing parameters
        num_reads = self.quantum_params['num_reads']
        annealing_time = self.quantum_params['annealing_time']
        
        # Simulate quantum annealing process
        best_energy = float('inf')
        best_solution = None
        energy_history = []
        
        for read in range(num_reads):
            # Simulate quantum tunneling and superposition
            current_solution = self._simulate_quantum_state()
            current_energy = self._calculate_energy(current_solution)
            
            energy_history.append(current_energy)
            
            if current_energy < best_energy:
                best_energy = current_energy
                best_solution = current_solution
        
        quantum_time = time.time() - start_time
        
        # Convert solution to assignment
        assignment = self._decode_solution(best_solution)
        
        quantum_results = {
            'solution': assignment,
            'energy': best_energy,
            'time': quantum_time,
            'energy_history': energy_history,
            'num_reads': num_reads,
            'annealing_time': annealing_time
        }
        
        self.quantum_solution = quantum_results
        
        print(f"Quantum annealing completed in {quantum_time:.3f} seconds")
        print(f"Best energy: {best_energy:.2f}")
        print(f"Solution quality: {len(assignment)} SKUs assigned")
        
        return quantum_results
    
    def _simulate_quantum_state(self) -> np.ndarray:
        """Simulate quantum state evolution"""
        
        num_vars = self.qubo_matrix.shape[0]
        
        # Start with random quantum state
        quantum_state = np.random.random(num_vars)
        
        # Simulate quantum annealing schedule
        temp_start = 10.0
        temp_end = 0.01
        num_steps = 100
        
        for step in range(num_steps):
            # Temperature following geometric schedule
            temp = temp_start * (temp_end / temp_start) ** (step / num_steps)
            
            # Quantum tunneling simulation
            for i in range(num_vars):
                # Calculate energy difference for flipping this bit
                current_energy = quantum_state[i] * self.qubo_matrix[i, i]
                for j in range(num_vars):
                    if i != j:
                        current_energy += quantum_state[i] * quantum_state[j] * self.qubo_matrix[i, j]
                
                # Flip with probability based on temperature
                flip_energy = -current_energy  # Energy after flip
                delta_e = flip_energy - current_energy
                
                if delta_e < 0 or np.random.random() < np.exp(-delta_e / temp):
                    quantum_state[i] = 1 - quantum_state[i]  # Flip bit
        
        # Convert to binary solution
        binary_solution = (quantum_state > 0.5).astype(int)
        
        return binary_solution
    
    def _calculate_energy(self, solution: np.ndarray) -> float:
        """Calculate energy of a quantum state"""
        
        energy = 0.0
        
        # Linear terms
        for i in range(len(solution)):
            energy += solution[i] * self.qubo_matrix[i, i]
        
        # Quadratic terms
        for i in range(len(solution)):
            for j in range(i+1, len(solution)):
                energy += solution[i] * solution[j] * self.qubo_matrix[i, j]
        
        return energy
    
    def _decode_solution(self, binary_solution: np.ndarray) -> Dict[int, int]:
        """Decode binary solution to SKU-location assignments"""
        
        assignment = {}
        num_locations = len(self.locations)
        
        for i, sku in enumerate(self.skus):
            # Find which location this SKU is assigned to
            for j in range(num_locations):
                var_idx = i * num_locations + j
                if binary_solution[var_idx] == 1:
                    assignment[sku['id']] = j + 1  # Location IDs are 1-based
                    break
        
        return assignment
    
    def classical_baseline(self) -> Dict[str, Any]:
        """Classical optimization baseline for comparison"""
        
        print("\n=== CLASSICAL BASELINE OPTIMIZATION ===")
        
        start_time = time.time()
        
        # Greedy assignment as baseline
        assignment = {}
        location_usage = {}
        
        # Sort SKUs by velocity (high to low)
        sorted_skus = sorted(self.skus, key=lambda x: x['velocity'], reverse=True)
        
        # Sort locations by accessibility (low to high)
        sorted_locations = sorted(self.locations, key=lambda x: x['accessibility'])
        
        for sku in sorted_skus:
            for location in sorted_locations:
                loc_id = location['id']
                if loc_id not in location_usage:
                    location_usage[loc_id] = {'space': 0.0, 'weight': 0.0}
                
                if (location_usage[loc_id]['space'] + sku['space_req'] <= location['capacity'] and
                    location_usage[loc_id]['weight'] + sku['weight'] <= location['weight_limit']):
                    
                    assignment[sku['id']] = loc_id
                    location_usage[loc_id]['space'] += sku['space_req']
                    location_usage[loc_id]['weight'] += sku['weight']
                    break
        
        classical_time = time.time() - start_time
        
        # Calculate objective value
        objective_value = self._calculate_objective(assignment)
        
        classical_results = {
            'solution': assignment,
            'objective': objective_value,
            'time': classical_time,
            'method': 'Greedy Heuristic'
        }
        
        self.classical_solution = classical_results
        
        print(f"Classical optimization completed in {classical_time:.3f} seconds")
        print(f"Objective value: {objective_value:.2f}")
        print(f"Solution quality: {len(assignment)} SKUs assigned")
        
        return classical_results
    
    def _calculate_objective(self, assignment: Dict[int, int]) -> float:
        """Calculate objective function value"""
        
        total_score = 0.0
        
        for sku_id, loc_id in assignment.items():
            sku = next(s for s in self.skus if s['id'] == sku_id)
            location = next(l for l in self.locations if l['id'] == loc_id)
            
            # Velocity-accessibility score
            velocity_score = sku['velocity'] / location['accessibility']
            
            # Zone preference
            zone_score = location['zone_preference'] * 10
            
            total_score += velocity_score + zone_score
        
        return total_score
    
    def hybrid_quantum_classical(self) -> Dict[str, Any]:
        """Hybrid quantum-classical optimization"""
        
        print("\n=== HYBRID QUANTUM-CLASSICAL OPTIMIZATION ===")
        
        start_time = time.time()
        
        # Step 1: Quantum annealing for global search
        quantum_results = self.quantum_annealing_simulation()
        quantum_assignment = quantum_results['solution']
        
        # Step 2: Classical local refinement
        refined_assignment = self._local_refinement(quantum_assignment)
        
        hybrid_time = time.time() - start_time
        
        # Calculate objective value
        objective_value = self._calculate_objective(refined_assignment)
        
        hybrid_results = {
            'solution': refined_assignment,
            'objective': objective_value,
            'time': hybrid_time,
            'quantum_energy': quantum_results['energy'],
            'method': 'Hybrid Quantum-Classical'
        }
        
        self.hybrid_solution = hybrid_results
        
        print(f"Hybrid optimization completed in {hybrid_time:.3f} seconds")
        print(f"Objective value: {objective_value:.2f}")
        print(f"Quantum energy: {quantum_results['energy']:.2f}")
        print(f"Solution quality: {len(refined_assignment)} SKUs assigned")
        
        return hybrid_results
    
    def _local_refinement(self, initial_assignment: Dict[int, int]) -> Dict[int, int]:
        """Local refinement of quantum solution"""
        
        assignment = initial_assignment.copy()
        improved = True
        iterations = 0
        
        while improved and iterations < 100:
            improved = False
            iterations += 1
            
            # Try local improvements
            for sku_id, current_loc_id in list(assignment.items()):
                sku = next(s for s in self.skus if s['id'] == sku_id)
                current_location = next(l for l in self.locations if l['id'] == current_loc_id)
                current_score = sku['velocity'] / current_location['accessibility']
                
                # Try better locations
                for location in self.locations:
                    if location['id'] != current_loc_id:
                        new_score = sku['velocity'] / location['accessibility']
                        
                        if new_score > current_score:
                            # Check if move is feasible
                            if self._is_move_feasible(sku_id, current_loc_id, location['id'], assignment):
                                assignment[sku_id] = location['id']
                                improved = True
                                break
        
        return assignment
    
    def _is_move_feasible(self, sku_id: int, from_loc_id: int, to_loc_id: int, assignment: Dict[int, int]) -> bool:
        """Check if moving SKU is feasible"""
        
        sku = next(s for s in self.skus if s['id'] == sku_id)
        to_location = next(l for l in self.locations if l['id'] == to_loc_id)
        
        # Calculate current usage of target location
        current_space = 0.0
        current_weight = 0.0
        
        for other_sku_id, other_loc_id in assignment.items():
            if other_sku_id != sku_id and other_loc_id == to_loc_id:
                other_sku = next(s for s in self.skus if s['id'] == other_sku_id)
                current_space += other_sku['space_req']
                current_weight += other_sku['weight']
        
        # Check constraints
        if current_space + sku['space_req'] > to_location['capacity']:
            return False
        
        if current_weight + sku['weight'] > to_location['weight_limit']:
            return False
        
        return True
    
    def quantum_inspired_algorithm(self) -> Dict[str, Any]:
        """Quantum-inspired optimization on classical hardware"""
        
        print("\n=== QUANTUM-INSPIRED OPTIMIZATION ===")
        
        start_time = time.time()
        
        # Simulated Quantum Approximate Optimization Algorithm (QAOA)
        num_layers = 3
        population_size = 50
        generations = 100
        
        # Initialize population
        population = [self._generate_random_solution() for _ in range(population_size)]
        
        best_solution = None
        best_energy = float('inf')
        energy_history = []
        
        for generation in range(generations):
            # Evaluate population
            energies = [self._calculate_energy(sol) for sol in population]
            
            # Track best solution
            min_energy_idx = np.argmin(energies)
            if energies[min_energy_idx] < best_energy:
                best_energy = energies[min_energy_idx]
                best_solution = population[min_energy_idx].copy()
            
            energy_history.append(best_energy)
            
            # Quantum-inspired evolution
            new_population = []
            
            for i in range(population_size):
                # Select parents (tournament selection)
                parent1 = self._tournament_selection(population, energies)
                parent2 = self._tournament_selection(population, energies)
                
                # Quantum-inspired crossover
                child = self._quantum_crossover(parent1, parent2)
                
                # Quantum-inspired mutation
                child = self._quantum_mutation(child)
                
                new_population.append(child)
            
            population = new_population
        
        quantum_inspired_time = time.time() - start_time
        
        # Decode best solution
        assignment = self._decode_solution(best_solution)
        objective_value = self._calculate_objective(assignment)
        
        quantum_inspired_results = {
            'solution': assignment,
            'objective': objective_value,
            'energy': best_energy,
            'time': quantum_inspired_time,
            'energy_history': energy_history,
            'method': 'Quantum-Inspired QAOA'
        }
        
        print(f"Quantum-inspired optimization completed in {quantum_inspired_time:.3f} seconds")
        print(f"Objective value: {objective_value:.2f}")
        print(f"Best energy: {best_energy:.2f}")
        print(f"Solution quality: {len(assignment)} SKUs assigned")
        
        return quantum_inspired_results
    
    def _generate_random_solution(self) -> np.ndarray:
        """Generate random binary solution"""
        
        num_vars = self.qubo_matrix.shape[0]
        solution = np.random.randint(0, 2, num_vars)
        
        # Ensure each SKU has exactly one location
        num_locations = len(self.locations)
        for i in range(len(self.skus)):
            # Clear all locations for this SKU
            for j in range(num_locations):
                solution[i * num_locations + j] = 0
            
            # Assign to random location
            random_loc = np.random.randint(0, num_locations)
            solution[i * num_locations + random_loc] = 1
        
        return solution
    
    def _tournament_selection(self, population: List[np.ndarray], energies: List[float]) -> np.ndarray:
        """Tournament selection for genetic algorithm"""
        
        tournament_size = 3
        candidates = np.random.choice(len(population), tournament_size, replace=False)
        
        best_idx = candidates[np.argmin([energies[i] for i in candidates])]
        return population[best_idx].copy()
    
    def _quantum_crossover(self, parent1: np.ndarray, parent2: np.ndarray) -> np.ndarray:
        """Quantum-inspired crossover operation"""
        
        child = np.zeros_like(parent1)
        
        # Quantum superposition-inspired crossover
        for i in range(len(parent1)):
            if np.random.random() < 0.5:
                child[i] = parent1[i]
            else:
                child[i] = parent2[i]
        
        # Ensure validity (each SKU has exactly one location)
        num_locations = len(self.locations)
        for i in range(len(self.skus)):
            sku_vars = child[i*num_locations:(i+1)*num_locations]
            if np.sum(sku_vars) != 1:
                # Fix by setting random location to 1
                for j in range(num_locations):
                    child[i*num_locations + j] = 0
                random_loc = np.random.randint(0, num_locations)
                child[i*num_locations + random_loc] = 1
        
        return child
    
    def _quantum_mutation(self, solution: np.ndarray) -> np.ndarray:
        """Quantum-inspired mutation operation"""
        
        mutated = solution.copy()
        mutation_rate = 0.1
        
        for i in range(len(self.skus)):
            if np.random.random() < mutation_rate:
                # Mutate SKU location
                num_locations = len(self.locations)
                
                # Clear current assignment
                for j in range(num_locations):
                    mutated[i*num_locations + j] = 0
                
                # Assign to new random location
                new_loc = np.random.randint(0, num_locations)
                mutated[i*num_locations + new_loc] = 1
        
        return mutated
    
    def compare_methods(self) -> Dict[str, Any]:
        """Compare all optimization methods"""
        
        print("\n=== METHOD COMPARISON ===")
        
        # Run all methods
        classical_results = self.classical_baseline()
        quantum_results = self.quantum_annealing_simulation()
        hybrid_results = self.hybrid_quantum_classical()
        quantum_inspired_results = self.quantum_inspired_algorithm()
        
        # Calculate improvements
        baseline_objective = classical_results['objective']
        
        quantum_improvement = ((quantum_results['solution'] and self._calculate_objective(quantum_results['solution']) or 0) - baseline_objective) / baseline_objective * 100
        hybrid_improvement = (hybrid_results['objective'] - baseline_objective) / baseline_objective * 100
        quantum_inspired_improvement = (quantum_inspired_results['objective'] - baseline_objective) / baseline_objective * 100
        
        comparison = {
            'classical': classical_results,
            'quantum': quantum_results,
            'hybrid': hybrid_results,
            'quantum_inspired': quantum_inspired_results,
            'improvements': {
                'quantum': quantum_improvement,
                'hybrid': hybrid_improvement,
                'quantum_inspired': quantum_inspired_improvement
            },
            'speedup': {
                'quantum_vs_classical': classical_results['time'] / quantum_results['time'] if quantum_results['time'] > 0 else 0,
                'hybrid_vs_classical': classical_results['time'] / hybrid_results['time'] if hybrid_results['time'] > 0 else 0,
                'quantum_inspired_vs_classical': classical_results['time'] / quantum_inspired_results['time'] if quantum_inspired_results['time'] > 0 else 0
            }
        }
        
        print(f"\nComparison Results:")
        print(f"Classical Objective: {baseline_objective:.2f}")
        print(f"Quantum Improvement: {quantum_improvement:.1f}%")
        print(f"Hybrid Improvement: {hybrid_improvement:.1f}%")
        print(f"Quantum-Inspired Improvement: {quantum_inspired_improvement:.1f}%")
        
        print(f"\nSpeedup Factors:")
        print(f"Quantum vs Classical: {comparison['speedup']['quantum_vs_classical']:.1f}x")
        print(f"Hybrid vs Classical: {comparison['speedup']['hybrid_vs_classical']:.1f}x")
        print(f"Quantum-Inspired vs Classical: {comparison['speedup']['quantum_inspired_vs_classical']:.1f}x")
        
        return comparison
    
    def visualize_quantum_results(self, comparison: Dict[str, Any]) -> None:
        """Visualize quantum optimization results"""
        
        print("\n=== QUANTUM OPTIMIZATION VISUALIZATION ===")
        
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        
        # 1. Objective value comparison
        methods = ['Classical', 'Quantum', 'Hybrid', 'Quantum-Inspired']
        objectives = [
            comparison['classical']['objective'],
            comparison['quantum']['solution'] and self._calculate_objective(comparison['quantum']['solution']) or 0,
            comparison['hybrid']['objective'],
            comparison['quantum_inspired']['objective']
        ]
        
        bars = axes[0, 0].bar(methods, objectives, color=['lightblue', 'lightgreen', 'lightyellow', 'lightcoral'])
        axes[0, 0].set_title('Objective Value Comparison')
        axes[0, 0].set_ylabel('Objective Value')
        axes[0, 0].grid(True, alpha=0.3)
        
        # Add value labels
        for bar, obj in zip(bars, objectives):
            height = bar.get_height()
            axes[0, 0].text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                           f'{obj:.1f}', ha='center', va='bottom')
        
        # 2. Execution time comparison
        times = [
            comparison['classical']['time'],
            comparison['quantum']['time'],
            comparison['hybrid']['time'],
            comparison['quantum_inspired']['time']
        ]
        
        axes[0, 1].bar(methods, times, color=['lightblue', 'lightgreen', 'lightyellow', 'lightcoral'])
        axes[0, 1].set_title('Execution Time Comparison')
        axes[0, 1].set_ylabel('Time (seconds)')
        axes[0, 1].grid(True, alpha=0.3)
        
        # 3. Improvement percentage
        improvements = [
            0,  # Classical baseline
            comparison['improvements']['quantum'],
            comparison['improvements']['hybrid'],
            comparison['improvements']['quantum_inspired']
        ]
        
        axes[0, 2].bar(methods, improvements, color=['lightblue', 'lightgreen', 'lightyellow', 'lightcoral'])
        axes[0, 2].set_title('Improvement Over Classical')
        axes[0, 2].set_ylabel('Improvement (%)')
        axes[0, 2].grid(True, alpha=0.3)
        
        # 4. Quantum energy evolution
        if 'energy_history' in comparison['quantum']:
            energy_history = comparison['quantum']['energy_history']
            axes[1, 0].plot(range(len(energy_history)), energy_history, 'g-', linewidth=2)
            axes[1, 0].set_title('Quantum Annealing Energy Evolution')
            axes[1, 0].set_xlabel('Iteration')
            axes[1, 0].set_ylabel('Energy')
            axes[1, 0].grid(True, alpha=0.3)
        
        # 5. Quantum-inspired convergence
        if 'energy_history' in comparison['quantum_inspired']:
            energy_history = comparison['quantum_inspired']['energy_history']
            axes[1, 1].plot(range(len(energy_history)), energy_history, 'b-', linewidth=2)
            axes[1, 1].set_title('Quantum-Inspired Convergence')
            axes[1, 1].set_xlabel('Generation')
            axes[1, 1].set_ylabel('Best Energy')
            axes[1, 1].grid(True, alpha=0.3)
        
        # 6. Speedup factors
        speedup_methods = ['Quantum', 'Hybrid', 'Quantum-Inspired']
        speedups = [
            comparison['speedup']['quantum_vs_classical'],
            comparison['speedup']['hybrid_vs_classical'],
            comparison['speedup']['quantum_inspired_vs_classical']
        ]
        
        axes[1, 2].bar(speedup_methods, speedups, color=['lightgreen', 'lightyellow', 'lightcoral'])
        axes[1, 2].set_title('Speedup Over Classical')
        axes[1, 2].set_ylabel('Speedup Factor')
        axes[1, 2].grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
    
    def analyze_quantum_advantage(self, comparison: Dict[str, Any]) -> Dict[str, Any]:
        """Analyze quantum advantage characteristics"""
        
        print("\n=== QUANTUM ADVANTAGE ANALYSIS ===")
        
        # Calculate quantum advantage metrics
        quantum_improvement = comparison['improvements']['quantum']
        hybrid_improvement = comparison['improvements']['hybrid']
        quantum_speedup = comparison['speedup']['quantum_vs_classical']
        
        # Determine if quantum advantage is achieved
        advantage_threshold = 10  # 10% improvement threshold
        speedup_threshold = 2   # 2x speedup threshold
        
        quantum_advantage = {
            'achieved': quantum_improvement > advantage_threshold or quantum_speedup > speedup_threshold,
            'improvement_advantage': quantum_improvement > advantage_threshold,
            'speedup_advantage': quantum_speedup > speedup_threshold,
            'hybrid_improvement': hybrid_improvement,
            'scaling_factor': len(self.skus) / 100,  # Normalized to 100 SKUs
        }
        
        print(f"Quantum Advantage Analysis:")
        print(f"  Quantum Improvement: {quantum_improvement:.1f}% (threshold: {advantage_threshold}%)")
        print(f"  Quantum Speedup: {quantum_speedup:.1f}x (threshold: {speedup_threshold}x)")
        print(f"  Hybrid Improvement: {hybrid_improvement:.1f}%")
        print(f"  Scaling Factor: {quantum_advantage['scaling_factor']:.1f}x")
        print(f"  Quantum Advantage Achieved: {quantum_advantage['achieved']}")
        
        # Predict quantum advantage for larger problems
        if quantum_advantage['achieved']:
            print(f"\nQuantum advantage detected!")
            print(f"Expected improvement for larger problems: {quantum_improvement * quantum_advantage['scaling_factor']:.1f}%")
        else:
            print(f"\nQuantum advantage not yet achieved.")
            print(f"Problem size may need to be larger for quantum advantage.")
        
        return quantum_advantage
```
