```python
# Digital Twin data structures
@dataclass
class IOSSensor:
    """IoT sensor for real-time data collection"""
    id: str
    type: str  # 'picker', 'conveyor', 'shelf', 'entrance'
    location: Tuple[float, float]
    status: str = 'active'
    last_update: datetime = field(default_factory=datetime.now)
    data_history: deque = field(default_factory=lambda: deque(maxlen=1000))

@dataclass
class PickerProfile:
    """Individual picker performance and fatigue model"""
    id: int
    name: str
    skill_level: float  # 0.0-1.0
    current_fatigue: float = 0.0  # 0.0-1.0
    orders_completed: int = 0
    total_distance: float = 0.0
    efficiency_history: List[float] = field(default_factory=list)
    break_schedule: List[datetime] = field(default_factory=list)

@dataclass
class EquipmentStatus:
    """Equipment monitoring and predictive maintenance"""
    id: str
    type: str  # 'conveyor', 'cart', 'scanner'
    location: Tuple[float, float]
    operational_status: str = 'running'
    utilization: float = 0.0
    maintenance_score: float = 1.0  # 1.0=perfect, 0.0=failure
    predicted_failure_time: datetime = None

@dataclass
class OrderFlowData:
    """Real-time order flow and demand patterns"""
    timestamp: datetime
    order_volume: int
    order_types: Dict[str, int]  # 'urgent', 'normal', 'bulk'
    peak_hours: bool = False
    seasonal_factor: float = 1.0
    external_factors: Dict[str, float] = field(default_factory=dict)

class DigitalTwinArchitecture:
    """Four-layer Digital Twin architecture for warehouse ecosystem"""
    
    def __init__(self, warehouse_size=(100, 150), num_zones=10):
        self.warehouse_size = warehouse_size
        self.num_zones = num_zones
        
        # Layer 1: Physical Layer - IoT sensors and equipment
        self.sensors = []
        self.rfid_tags = []
        self.picker_wearables = []
        self.equipment = []
        
        # Layer 2: Connectivity Layer - Data transmission
        self.data_streams = defaultdict(list)
        self.network_status = 'optimal'
        self.data_latency = 0.045  # seconds
        
        # Layer 3: Data Layer - Storage and processing
        self.time_series_db = defaultdict(deque)
        self.graph_db = {}  # Warehouse topology
        self.real_time_data = {}
        
        # Layer 4: Application Layer - Analytics and optimization
        self.ml_models = {}
        self.optimization_engine = None
        self.dashboard_data = {}
        
        # Simulation state
        self.current_time = datetime.now()
        self.simulation_speed = 1.0  # Real-time factor
        
        # Performance metrics
        self.kpi_history = defaultdict(list)
        self.alert_system = []
        
        self._initialize_physical_layer()
    
    def _initialize_physical_layer(self):
        """Initialize IoT sensors, equipment, and picker profiles"""
        # Create IoT sensors (847 sensors as per source example)
        sensor_types = ['picker', 'conveyor', 'shelf', 'entrance', 'exit', 'charging']
        for i in range(847):
            sensor_type = np.random.choice(sensor_types)
            location = (np.random.uniform(0, self.warehouse_size[0]), 
                       np.random.uniform(0, self.warehouse_size[1]))
            sensor = IOSSensor(
                id=f"sensor_{i}",
                type=sensor_type,
                location=location
            )
            self.sensors.append(sensor)
        
        # Create RFID tags (12,500 as per source)
        for i in range(12500):
            self.rfid_tags.append({
                'id': f"rfid_{i}",
                'status': 'active',
                'last_scan': self.current_time
            })
        
        # Create picker wearables and profiles (156 pickers)
        for i in range(156):
            wearable = {
                'id': f"wearable_{i}",
                'picker_id': i,
                'status': 'active',
                'battery_level': np.random.uniform(0.3, 1.0),
                'location': (0, 0)
            }
            self.picker_wearables.append(wearable)
            
            # Create corresponding picker profile
            profile = PickerProfile(
                id=i,
                name=f"Picker_{i}",
                skill_level=np.random.uniform(0.6, 1.0),
                current_fatigue=np.random.uniform(0.0, 0.3)
            )
            self.real_time_data[f"picker_{i}"] = profile
        
        # Create equipment
        equipment_types = ['conveyor', 'cart', 'scanner', 'charger']
        for i in range(50):
            eq_type = np.random.choice(equipment_types)
            location = (np.random.uniform(0, self.warehouse_size[0]), 
                       np.random.uniform(0, self.warehouse_size[1]))
            equipment = EquipmentStatus(
                id=f"eq_{i}",
                type=eq_type,
                location=location,
                maintenance_score=np.random.uniform(0.7, 1.0)
            )
            self.equipment.append(equipment)
    
    def collect_sensor_data(self):
        """Collect real-time data from all IoT sensors"""
        timestamp = self.current_time
        
        for sensor in self.sensors:
            if sensor.status == 'active':
                # Simulate sensor data based on type
                if sensor.type == 'picker':
                    data_value = np.random.normal(0.8, 0.1)  # Picker activity level
                elif sensor.type == 'conveyor':
                    data_value = np.random.normal(15.0, 2.0)  # Conveyor speed m/min
                elif sensor.type == 'shelf':
                    data_value = np.random.uniform(0.2, 0.9)  # Shelf fill level
                elif sensor.type == 'entrance':
                    data_value = np.random.poisson(5)  # Orders entering per minute
                else:
                    data_value = np.random.normal(0.5, 0.1)  # Generic sensor
                
                sensor_data = {
                    'timestamp': timestamp,
                    'value': data_value,
                    'sensor_id': sensor.id,
                    'type': sensor.type,
                    'location': sensor.location
                }
                
                sensor.data_history.append(sensor_data)
                self.data_streams[sensor.type].append(sensor_data)
                sensor.last_update = timestamp
    
    def update_picker_performance(self):
        """Update picker performance metrics and fatigue models"""
        for wearable in self.picker_wearables:
            picker_id = wearable['picker_id']
            profile_key = f"picker_{picker_id}"
            
            if profile_key in self.real_time_data:
                profile = self.real_time_data[profile_key]
                
                # Update fatigue based on work duration
                hours_worked = (self.current_time.hour - 6) % 8  # 8-hour shifts
                fatigue_increase = hours_worked * 0.05  # 5% fatigue per hour
                profile.current_fatigue = min(1.0, profile.current_fatigue + fatigue_increase * 0.01)
                
                # Update efficiency based on fatigue and skill
                efficiency = profile.skill_level * (1 - profile.current_fatigue * 0.5)
                profile.efficiency_history.append(efficiency)
                
                # Update battery level
                wearable['battery_level'] -= np.random.uniform(0.001, 0.005)
                if wearable['battery_level'] < 0.2:
                    self.alert_system.append({
                        'timestamp': self.current_time,
                        'type': 'low_battery',
                        'picker_id': picker_id,
                        'battery_level': wearable['battery_level']
                    })
    
    def predict_equipment_failure(self):
        """Predictive maintenance for equipment"""
        for equipment in self.equipment:
            # Simulate degradation
            degradation = np.random.normal(0, 0.001)
            equipment.maintenance_score = max(0.0, equipment.maintenance_score - degradation)
            
            # Predict failure if score is low
            if equipment.maintenance_score < 0.3:
                hours_to_failure = np.random.exponential(24)  # Exponential distribution
                equipment.predicted_failure_time = self.current_time + timedelta(hours=hours_to_failure)
                
                self.alert_system.append({
                    'timestamp': self.current_time,
                    'type': 'maintenance_alert',
                    'equipment_id': equipment.id,
                    'maintenance_score': equipment.maintenance_score,
                    'predicted_failure': equipment.predicted_failure_time
                })
    
    def generate_order_flow_forecast(self, hours_ahead=4):
        """Generate order flow predictions using ML models"""
        # Simulate LSTM + CNN ensemble prediction
        base_volume = np.random.poisson(50)  # Base order volume
        
        # Time-based patterns
        hour_factor = 1.0
        if 10 <= self.current_time.hour <= 12:  # Peak morning
            hour_factor = 1.8
        elif 14 <= self.current_time.hour <= 16:  # Peak afternoon
            hour_factor = 1.6
        elif 20 <= self.current_time.hour <= 22:  # Evening peak
            hour_factor = 1.4
        
        # Seasonal factors
        day_of_week = self.current_time.weekday()
        if day_of_week >= 5:  # Weekend
            seasonal_factor = 1.3
        else:
            seasonal_factor = 1.0
        
        # External factors (weather, promotions, etc.)
        weather_factor = np.random.uniform(0.8, 1.2)
        promotion_factor = 1.5 if np.random.random() < 0.1 else 1.0  # 10% chance of promotion
        
        predicted_volume = int(base_volume * hour_factor * seasonal_factor * 
                              weather_factor * promotion_factor)
        
        # Create order flow data
        order_flow = OrderFlowData(
            timestamp=self.current_time,
            order_volume=predicted_volume,
            order_types={
                'urgent': int(predicted_volume * 0.15),
                'normal': int(predicted_volume * 0.70),
                'bulk': int(predicted_volume * 0.15)
            },
            peak_hours=hour_factor > 1.2,
            seasonal_factor=seasonal_factor,
            external_factors={
                'weather': weather_factor,
                'promotion': promotion_factor
            }
        )
        
        return order_flow

print("Digital Twin architecture defined successfully")
```

```python
# Dynamic Optimization Engine
class DynamicOptimizationEngine:
    """Real-time multi-objective optimization for order batching"""
    
    def __init__(self, digital_twin):
        self.digital_twin = digital_twin
        self.optimization_interval = 45  # seconds (as per source)
        self.last_optimization = None
        
        # Optimization parameters
        self.alpha = 0.4  # Weight for total picking time
        self.beta = 0.3   # Weight for labor costs
        self.gamma = 0.3  # Weight for delay penalties
        
        # Current batching solution
        self.current_batches = {}
        self.batch_assignments = {}
        
        # Performance tracking
        self.optimization_history = []
        self.kpi_trends = defaultdict(list)
    
    def calculate_objective_function(self, batch_solution):
        """Multi-objective function for real-time optimization"""
        # Extract picker assignments from solution
        picker_assignments = batch_solution.get('picker_assignments', {})
        
        # Calculate total picking time
        total_picking_time = 0
        for picker_id, batch_list in picker_assignments.items():
            for batch in batch_list:
                # Simulate picking time calculation
                batch_time = self._estimate_batch_picking_time(batch, picker_id)
                total_picking_time += batch_time
        
        # Calculate labor costs
        labor_costs = self._calculate_labor_costs(picker_assignments)
        
        # Calculate delay penalties
        delay_penalties = self._calculate_delay_penalties(batch_solution)
        
        # Multi-objective combination
        total_objective = (self.alpha * total_picking_time + 
                          self.beta * labor_costs + 
                          self.gamma * delay_penalties)
        
        return {
            'total_objective': total_objective,
            'picking_time': total_picking_time,
            'labor_costs': labor_costs,
            'delay_penalties': delay_penalties,
            'breakdown': {
                'time_weight': self.alpha * total_picking_time,
                'cost_weight': self.beta * labor_costs,
                'delay_weight': self.gamma * delay_penalties
            }
        }
    
    def _estimate_batch_picking_time(self, batch, picker_id):
        """Estimate picking time for a batch considering picker performance"""
        profile_key = f"picker_{picker_id}"
        
        # Get picker efficiency
        if profile_key in self.digital_twin.real_time_data:
            profile = self.digital_twin.real_time_data[profile_key]
            efficiency = profile.skill_level * (1 - profile.current_fatigue * 0.5)
        else:
            efficiency = 0.8  # Default efficiency
        
        # Base picking time (simplified calculation)
        base_time = len(batch) * 2.0  # 2 minutes per order average
        
        # Adjust for efficiency
        adjusted_time = base_time / efficiency
        
        # Add travel time component
        travel_time = len(batch) * 1.5  # 1.5 minutes travel per order
        
        return adjusted_time + travel_time
    
    def _calculate_labor_costs(self, picker_assignments):
        """Calculate labor costs based on picker utilization and fatigue"""
        total_cost = 0
        
        for picker_id, batch_list in picker_assignments.items():
            profile_key = f"picker_{picker_id}"
            
            if profile_key in self.digital_twin.real_time_data:
                profile = self.digital_twin.real_time_data[profile_key]
                
                # Base cost per hour
                base_hourly_rate = 25.0  # $25/hour
                
                # Adjust for skill level
                skill_adjustment = 1.0 + (profile.skill_level - 0.7) * 0.5
                
                # Calculate total work time
                total_work_time = sum(self._estimate_batch_picking_time(batch, picker_id) 
                                    for batch in batch_list) / 60.0  # Convert to hours
                
                # Fatigue penalty (higher cost for fatigued workers)
                fatigue_penalty = 1.0 + profile.current_fatigue * 0.2
                
                picker_cost = base_hourly_rate * skill_adjustment * total_work_time * fatigue_penalty
                total_cost += picker_cost
        
        return total_cost
    
    def _calculate_delay_penalties(self, batch_solution):
        """Calculate penalties for order delays and service level violations"""
        total_penalty = 0
        
        # Get order deadlines from solution
        order_deadlines = batch_solution.get('order_deadlines', {})
        completion_times = batch_solution.get('completion_times', {})
        
        for order_id, deadline in order_deadlines.items():
            if order_id in completion_times:
                completion_time = completion_times[order_id]
                delay = completion_time - deadline
                
                if delay > 0:  # Order is delayed
                    # Penalty increases with delay severity
                    penalty = delay * 10.0  # $10 per minute of delay
                    total_penalty += penalty
        
        return total_penalty
    
    def optimize_batching(self, current_orders, picker_availability):
        """Run real-time optimization with current constraints"""
        # Simulate optimization process
        start_time = time.time()
        
        # Generate initial solution (simplified heuristic)
        solution = self._generate_initial_solution(current_orders, picker_availability)
        
        # Evaluate objective function
        objectives = self.calculate_objective_function(solution)
        
        # Simulate improvement iterations
        best_solution = solution
        best_objectives = objectives
        
        for iteration in range(10):  # Simplified optimization loop
            # Generate neighbor solution
            neighbor_solution = self._generate_neighbor_solution(best_solution)
            neighbor_objectives = self.calculate_objective_function(neighbor_solution)
            
            # Accept if better
            if neighbor_objectives['total_objective'] < best_objectives['total_objective']:
                best_solution = neighbor_solution
                best_objectives = neighbor_objectives
        
        optimization_time = time.time() - start_time
        
        # Store results
        self.current_batches = best_solution
        self.last_optimization = self.digital_twin.current_time
        
        # Record optimization history
        self.optimization_history.append({
            'timestamp': self.digital_twin.current_time,
            'optimization_time': optimization_time,
            'objective_value': best_objectives['total_objective'],
            'picking_time': best_objectives['picking_time'],
            'labor_costs': best_objectives['labor_costs'],
            'delay_penalties': best_objectives['delay_penalties']
        })
        
        return best_solution, best_objectives
    
    def _generate_initial_solution(self, orders, pickers):
        """Generate initial feasible solution"""
        # Simple greedy assignment
        picker_assignments = {picker_id: [] for picker_id in pickers}
        picker_loads = {picker_id: 0 for picker_id in pickers}
        
        # Sort orders by priority
        sorted_orders = sorted(orders, key=lambda x: x.get('priority', 1), reverse=True)
        
        for order in sorted_orders:
            # Find least loaded picker
            best_picker = min(pickers, key=lambda x: picker_loads[x])
            
            # Create batch if needed
            if not picker_assignments[best_picker] or picker_loads[best_picker] > 20:
                picker_assignments[best_picker].append([order])
            else:
                picker_assignments[best_picker][-1].append(order)
            
            picker_loads[best_picker] += order.get('volume', 1)
        
        return {
            'picker_assignments': picker_assignments,
            'order_deadlines': {order['id']: order.get('deadline', 100) for order in orders},
            'completion_times': {order['id']: 50 for order in orders}  # Simplified
        }
    
    def _generate_neighbor_solution(self, current_solution):
        """Generate neighbor solution by small modification"""
        # Copy current solution
        new_solution = {
            'picker_assignments': {},
            'order_deadlines': current_solution['order_deadlines'].copy(),
            'completion_times': current_solution['completion_times'].copy()
        }
        
        # Copy assignments
        for picker_id, batches in current_solution['picker_assignments'].items():
            new_solution['picker_assignments'][picker_id] = [batch.copy() for batch in batches]
        
        # Random modification
        if np.random.random() < 0.5 and len(new_solution['picker_assignments']) > 1:
            # Move a batch from one picker to another
            pickers = list(new_solution['picker_assignments'].keys())
            source_picker = np.random.choice(pickers)
            target_picker = np.random.choice([p for p in pickers if p != source_picker])
            
            if new_solution['picker_assignments'][source_picker]:
                batch = new_solution['picker_assignments'][source_picker].pop()
                new_solution['picker_assignments'][target_picker].append(batch)
        
        return new_solution

print("Dynamic optimization engine defined")
```

```python
try:
    try:
        # Simulate Digital Twin operation
        def simulate_digital_twin_operation(digital_twin, opt_engine, duration_hours=8):
            """Simulate Digital Twin operation for specified duration"""
            print(f"=== Simulating Digital Twin Operation ({duration_hours} hours) ===")
            
            # Simulation parameters
            time_step_minutes = 5  # 5-minute time steps
            total_steps = (duration_hours * 60) // time_step_minutes
            
            # Performance tracking
            performance_metrics = {
                'timestamps': [],
                'order_volume': [],
                'picking_time': [],
                'labor_costs': [],
                'delay_penalties': [],
                'prediction_accuracy': [],
                'equipment_alerts': [],
                'picker_alerts': []
                'data_volume': []
            }
            
            print(f"Starting simulation with {total_steps} time steps...")
            
            for step in range(total_steps):
                # Update simulation time
                digital_twin.current_time += timedelta(minutes=time_step_minutes)
                current_time = digital_twin.current_time
                
                # Layer 1: Collect sensor data
                digital_twin.collect_sensor_data()
                
                # Update picker performance
                digital_twin.update_picker_performance()
                
                # Predictive maintenance
                digital_twin.predict_equipment_failure()
                
                # Generate order flow forecast
                order_forecast = digital_twin.generate_order_flow_forecast()
                
                # Create current orders list
                current_orders = []
                for i in range(order_forecast.order_volume):
                    order = {
                        'id': f"order_{step}_{i}",
                        'volume': np.random.randint(1, 5),
                        'priority': np.random.choice(['urgent', 'normal', 'bulk'], 
                                              p=[0.15, 0.70, 0.15]),
                        'deadline': np.random.randint(30, 120),
                        'arrival_time': current_time
                    }
                    current_orders.append(order)
                
                # Get available pickers
                available_pickers = []
                for wearable in digital_twin.picker_wearables:
                    if wearable['battery_level'] > 0.3 and wearable['status'] == 'active':
                        available_pickers.append(wearable['picker_id'])
                
                # Run optimization every 45 seconds (simulation time)
                if step % (45 // time_step_minutes) == 0:
                    solution, objectives = opt_engine.optimize_batching(current_orders, available_pickers)
                    
                    # Record performance metrics
                    performance_metrics['timestamps'].append(current_time)
                    performance_metrics['order_volume'].append(len(current_orders))
                    performance_metrics['picking_time'].append(objectives['picking_time'])
                    performance_metrics['labor_costs'].append(objectives['labor_costs'])
                    performance_metrics['delay_penalties'].append(objectives['delay_penalties'])
                    
                    # Calculate prediction accuracy (simulated)
                    actual_volume = len(current_orders)
                    predicted_volume = order_forecast.order_volume
                    accuracy = 1.0 - abs(actual_volume - predicted_volume) / max(actual_volume, predicted_volume)
                    performance_metrics['prediction_accuracy'].append(accuracy)
                    
                    # Count alerts
                    equipment_alerts = len([a for a in digital_twin.alert_system 
                                         if a['type'] == 'maintenance_alert' and 
                                            a['timestamp'] > current_time - timedelta(minutes=5)])
                    picker_alerts = len([a for a in digital_twin.alert_system 
                                       if a['type'] == 'low_battery' and 
                                          a['timestamp'] > current_time - timedelta(minutes=5)])
                    
                    performance_metrics['equipment_alerts'].append(equipment_alerts)
                    performance_metrics['picker_alerts'].append(picker_alerts)
                    
                    # Calculate data volume (simulated 2.3TB daily)
                    data_volume_mb = (2300 * 1024 * 1024) / (24 * 60) * time_step_minutes  # MB per time step
                    performance_metrics['data_volume'].append(data_volume_mb)
                
                # Progress reporting
                if step % 20 == 0:
                    progress = (step / total_steps) * 100
                    print(f"Simulation progress: {progress:.1f}% (Step {step}/{total_steps})")
                    print(f"  Current time: {current_time.strftime('%H:%M')}")
                    print(f"  Orders this period: {len(current_orders)}")
                    print(f"  Available pickers: {len(available_pickers)}")
                    print(f"  Active alerts: {len(digital_twin.alert_system)}")
            
            print(f"\n✅ Simulation completed!")
            return performance_metrics
        
        # Run simulation
        performance_metrics = simulate_digital_twin_operation(digital_twin, opt_engine, duration_hours=8)
    except Exception as e:
        print(f'Error in cell: {e}')
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    try:
        # What-if scenario analysis
        def simulate_what_if_scenarios(digital_twin, opt_engine):
            """Simulate different what-if scenarios"""
            print("=== What-If Scenario Analysis ===")
            
            scenarios = {
                'order_surge': {'volume_multiplier': 2.3, 'description': '340% order volume increase (Black Friday)'},
                'equipment_failure': {'equipment_availability': 0.7, 'description': '30% equipment failure'},
                'picker_shortage': {'picker_availability': 0.6, 'description': '40% picker shortage'},
                'normal': {'volume_multiplier': 1.0, 'equipment_availability': 1.0, 'picker_availability': 1.0, 'description': 'Normal operations'}
            }
            
            scenario_results = {}
            
            for scenario_name, params in scenarios.items():
                print(f"\n📋 Testing Scenario: {params['description']}")
                
                # Reset time
                original_time = digital_twin.current_time
                
                # Simulate 2-hour period under scenario conditions
                scenario_metrics = {
                    'picking_time': [],
                    'completion_rate': [],
                    'delay_penalties': []
                    'adaptation_time': None
                }
                
                for minute in range(120):  # 2 hours
                    digital_twin.current_time += timedelta(minutes=1)
                    
                    # Apply scenario conditions
                    volume_multiplier = params.get('volume_multiplier', 1.0)
                    equipment_availability = params.get('equipment_availability', 1.0)
                    picker_availability = params.get('picker_availability', 1.0)
                    
                    # Generate orders based on scenario
                    base_volume = np.random.poisson(50)
                    order_volume = int(base_volume * volume_multiplier)
                    
                    current_orders = []
                    for i in range(min(order_volume, 200)):  # Cap for performance
                        order = {
                            'id': f"order_{scenario_name}_{minute}_{i}",
                            'volume': np.random.randint(1, 5),
                            'priority': np.random.choice(['urgent', 'normal', 'bulk'], p=[0.2, 0.6, 0.2]),
                            'deadline': np.random.randint(30, 120)
                        }
                        current_orders.append(order)
                    
                    # Get available resources under scenario
                    available_pickers = []
                    for wearable in digital_twin.picker_wearables:
                        if (wearable['battery_level'] > 0.3 and 
                            wearable['status'] == 'active' and 
                            np.random.random() < picker_availability):
                            available_pickers.append(wearable['picker_id'])
                    
                    # Run optimization
                    if current_orders and available_pickers:
                        solution, objectives = opt_engine.optimize_batching(current_orders, available_pickers)
                        
                        # Calculate completion rate
                        completion_rate = len(current_orders) / max(1, order_volume)
                        
                        scenario_metrics['picking_time'].append(objectives['picking_time'])
                        scenario_metrics['completion_rate'].append(completion_rate)
                        scenario_metrics['delay_penalties'].append(objectives['delay_penalties'])
                        
                        # Track adaptation time (time to reach 90% of normal performance)
                        if scenario_metrics['adaptation_time'] is None:
                            if completion_rate > 0.9 * picker_availability:
                                scenario_metrics['adaptation_time'] = minute
                
                # Calculate scenario averages
                avg_picking_time = np.mean(scenario_metrics['picking_time']) if scenario_metrics['picking_time'] else 0
                avg_completion_rate = np.mean(scenario_metrics['completion_rate']) if scenario_metrics['completion_rate'] else 0
                avg_delay_penalties = np.mean(scenario_metrics['delay_penalties']) if scenario_metrics['delay_penalties'] else 0
                adaptation_time = scenario_metrics['adaptation_time']
                
                scenario_results[scenario_name] = {
                    'avg_picking_time': avg_picking_time,
                    'avg_completion_rate': avg_completion_rate,
                    'avg_delay_penalties': avg_delay_penalties,
                    'adaptation_time': adaptation_time
                }
                
                print(f"  Average picking time: {avg_picking_time:.2f} minutes")
                print(f"  Average completion rate: {avg_completion_rate:.1%}")
                print(f"  Average delay penalties: ${avg_delay_penalties:.2f}")
                print(f"  Adaptation time: {adaptation_time} minutes" if adaptation_time else "  Adaptation time: N/A")
                
                # Restore time
                digital_twin.current_time = original_time
            
            return scenario_results
        
        # Run what-if analysis
        scenario_results = simulate_what_if_scenarios(digital_twin, opt_engine)
        
        # Compare with normal operations
        normal_picking_time = scenario_results['normal']['avg_picking_time']
        normal_completion = scenario_results['normal']['avg_completion_rate']
        
        print(f"\n🎯 Scenario Impact Analysis:")
        for scenario, results in scenario_results.items():
            if scenario != 'normal':
                time_impact = ((results['avg_picking_time'] - normal_picking_time) / normal_picking_time) * 100
                completion_impact = ((results['avg_completion_rate'] - normal_completion) / normal_completion) * 100
                
                print(f"  {scenario.replace('_', ' ').title()}:")
                print(f"    Time impact: {time_impact:+.1f}%")
                print(f"    Completion impact: {completion_impact:+.1f}%")
                print(f"    Adaptation time: {results['adaptation_time']} minutes" if results['adaptation_time'] else "    Adaptation time: N/A")
                print()
        
        # Black Friday performance
        black_friday_completion = scenario_results['order_surge']['avg_completion_rate']
        print(f"📈 Black Friday Performance: {black_friday_completion:.1%} on-time fulfillment")
        print(f"   (vs 78.2% for traditional static systems as per source)")
    except Exception as e:
        print(f'Error in cell: {e}')
except Exception as e:
    print(f'Error in cell: {e}')
```
