```python
class CustomerSegment(Enum):
    PRICE_SENSITIVE = "price_sensitive"
    QUALITY_FOCUSED = "quality_focused"
    CONVENIENCE_SEEKER = "convenience_seeker"
    LOYAL_PREMIUM = "loyal_premium"
    OCCASIONAL_USER = "occasional_user"

@dataclass
class CustomerProfile:
    """Customer profile with behavioral characteristics"""
    segment: CustomerSegment
    price_elasticity: float  # Price sensitivity coefficient
    time_sensitivity: float  # Time window flexibility
    loyalty_score: float  # Customer loyalty (0-1)
    order_frequency: float  # Orders per month
    avg_order_value: float  # Average order value
    nudge_susceptibility: float  # Response to behavioral nudges
    preferred_hours: List[int]  # Preferred delivery hours

@dataclass
class PricingStrategy:
    """Dynamic pricing strategy for demand shaping"""
    base_price: float
    peak_premium: float  # Premium during peak hours
    off_peak_discount: float  # Discount during off-peak hours
    loyalty_discount: float  # Discount for loyal customers
    bulk_discount: float  # Discount for large orders
    segment_pricing: Dict[CustomerSegment, float]  # Segment-specific adjustments

@dataclass
class BehavioralNudge:
    """Behavioral nudge for customer behavior modification"""
    nudge_type: str
    target_segment: CustomerSegment
    incentive_value: float
    psychological_trigger: str
    effectiveness: float  # Expected effectiveness (0-1)
    duration_days: int

@dataclass
class PartnershipProgram:
    """Strategic partnership program"""
    program_id: str
    partner_type: str  # restaurant, corporate, loyalty, etc.
    partner_name: str
    benefits: List[str]
    commission_rate: float
    customer_acquisition_rate: float
    retention_improvement: float

@dataclass
class MarketEcosystem:
    """Complete market ecosystem for demand shaping"""
    customer_segments: Dict[CustomerSegment, List[CustomerProfile]]
    pricing_strategy: PricingStrategy
    behavioral_nudges: List[BehavioralNudge]
    partnership_programs: List[PartnershipProgram]
    market_metrics: Dict
    competitive_position: float
    market_share: float
    profit_margin: float
```

```python
try:
    try:
        class DemandShapingOptimizer:
            """Optimizer for proactive demand shaping and market influence"""
            
            def __init__(self):
                self.market_size = 10000  # Total potential customers
               .current_penetration = 0.15  # Current market penetration
               .target_penetration = 0.35  # Target market penetration
                
            def create_market_ecosystem(self) -> MarketEcosystem:
                """Create comprehensive market ecosystem"""
                
                # Create customer segments
                customer_segments = self._create_customer_segments()
                
                # Create pricing strategy
                pricing_strategy = self._create_pricing_strategy()
                
                # Create behavioral nudges
                behavioral_nudges = self._create_behavioral_nudges()
                
                # Create partnership programs
                partnership_programs = self._create_partnership_programs()
                
                # Initialize market metrics
                market_metrics = {
                    'total_customers': self.market_size,
                    'active_customers': int(self.market_size * self.current_penetration),
                    'monthly_orders': 0,
                    'avg_order_value': 0,
                    'peak_hour_demand': 0.4,  # 40% of demand in peak hours
                    'off_peak_demand': 0.6,  # 60% in off-peak hours
                    'customer_acquisition_cost': 15.0,
                    'customer_lifetime_value': 450.0
                }
                
                return MarketEcosystem(
                    customer_segments=customer_segments,
                    pricing_strategy=pricing_strategy,
                    behavioral_nudges=behavioral_nudges,
                    partnership_programs=partnership_programs,
                    market_metrics=market_metrics,
                    competitive_position=0.25,  # 25% market position
                    market_share=self.current_penetration,
                    profit_margin=0.12  # 12% profit margin
                )
            
            def _create_customer_segments(self) -> Dict[CustomerSegment, List[CustomerProfile]]:
                """Create customer segments with detailed profiles"""
                
                segments = {}
                
                # Price-sensitive segment (35% of market)
                price_sensitive_profiles = []
                for i in range(350):
                    profile = CustomerProfile(
                        segment=CustomerSegment.PRICE_SENSITIVE,
                        price_elasticity=np.random.normal(-0.15, 0.05),  # High price sensitivity
                        time_sensitivity=np.random.uniform(0.2, 0.6),
                        loyalty_score=np.random.uniform(0.1, 0.4),
                        order_frequency=np.random.uniform(2, 6),
                        avg_order_value=np.random.uniform(15, 35),
                        nudge_susceptibility=np.random.uniform(0.6, 0.9),
                        preferred_hours=[11, 12, 13, 14, 18, 19]
                    )
                    price_sensitive_profiles.append(profile)
                segments[CustomerSegment.PRICE_SENSITIVE] = price_sensitive_profiles
                
                # Quality-focused segment (25% of market)
                quality_profiles = []
                for i in range(250):
                    profile = CustomerProfile(
                        segment=CustomerSegment.QUALITY_FOCUSED,
                        price_elasticity=np.random.normal(-0.05, 0.03),  # Low price sensitivity
                        time_sensitivity=np.random.uniform(0.3, 0.7),
                        loyalty_score=np.random.uniform(0.4, 0.7),
                        order_frequency=np.random.uniform(4, 8),
                        avg_order_value=np.random.uniform(35, 60),
                        nudge_susceptibility=np.random.uniform(0.3, 0.6),
                        preferred_hours=[12, 13, 19, 20, 21]
                    )
                    quality_profiles.append(profile)
                segments[CustomerSegment.QUALITY_FOCUSED] = quality_profiles
                
                # Convenience seeker segment (20% of market)
                convenience_profiles = []
                for i in range(200):
                    profile = CustomerProfile(
                        segment=CustomerSegment.CONVENIENCE_SEEKER,
                        price_elasticity=np.random.normal(-0.08, 0.04),
                        time_sensitivity=np.random.uniform(0.6, 0.9),  # High time sensitivity
                        loyalty_score=np.random.uniform(0.3, 0.6),
                        order_frequency=np.random.uniform(6, 12),
                        avg_order_value=np.random.uniform(25, 45),
                        nudge_susceptibility=np.random.uniform(0.5, 0.8),
                        preferred_hours=[11, 12, 18, 19, 20]
                    )
                    convenience_profiles.append(profile)
                segments[CustomerSegment.CONVENIENCE_SEEKER] = convenience_profiles
                
                # Loyal premium segment (15% of market)
                loyal_profiles = []
                for i in range(150):
                    profile = CustomerProfile(
                        segment=CustomerSegment.LOYAL_PREMIUM,
                        price_elasticity=np.random.normal(-0.03, 0.02),  # Very low price sensitivity
                        time_sensitivity=np.random.uniform(0.4, 0.8),
                        loyalty_score=np.random.uniform(0.7, 0.95),  # High loyalty
                        order_frequency=np.random.uniform(8, 15),
                        avg_order_value=np.random.uniform(45, 80),
                        nudge_susceptibility=np.random.uniform(0.2, 0.5),
                        preferred_hours=[12, 13, 19, 20, 21]
                    )
                    loyal_profiles.append(profile)
                segments[CustomerSegment.LOYAL_PREMIUM] = loyal_profiles
                
                # Occasional user segment (5% of market)
                occasional_profiles = []
                for i in range(50):
                    profile = CustomerProfile(
                        segment=CustomerSegment.OCCASIONAL_USER,
                        price_elasticity=np.random.normal(-0.12, 0.06),
                        time_sensitivity=np.random.uniform(0.1, 0.5),
                        loyalty_score=np.random.uniform(0.05, 0.25),
                        order_frequency=np.random.uniform(1, 3),
                        avg_order_value=np.random.uniform(20, 40),
                        nudge_susceptibility=np.random.uniform(0.7, 1.0),
                        preferred_hours=[11, 12, 18, 19, 20]
                    )
                    occasional_profiles.append(profile)
                segments[CustomerSegment.OCCASIONAL_USER] = occasional_profiles
                
                return segments
            
            def _create_pricing_strategy(self) -> PricingStrategy:
                """Create dynamic pricing strategy"""
                
                segment_pricing = {
                    CustomerSegment.PRICE_SENSITIVE: -0.15,  # 15% discount
                    CustomerSegment.QUALITY_FOCUSED: 0.05,   # 5% premium
                    CustomerSegment.CONVENIENCE_SEEKER: 0.02,  # 2% premium
                    CustomerSegment.LOYAL_PREMIUM: -0.10,     # 10% discount
                    CustomerSegment.OCCASIONAL_USER: -0.08   # 8% discount
                }
                
                return PricingStrategy(
                    base_price=25.0,
                    peak_premium=0.20,  # 20% premium during peak hours
                    off_peak_discount=0.15,  # 15% discount during off-peak
                    loyalty_discount=0.10,  # 10% loyalty discount
                    bulk_discount=0.05,  # 5% bulk order discount
                    segment_pricing=segment_pricing
                )
            
            def _create_behavioral_nudges(self) -> List[BehavioralNudge]:
                """Create behavioral nudges for demand shaping"""
                
                nudges = [
                    BehavioralNudge(
                        nudge_type="off_peak_gamification",
                        target_segment=CustomerSegment.PRICE_SENSITIVE,
                        incentive_value=0.15,  # 15% off-peak discount
                        psychological_trigger="achievement_and_rewards",
                        effectiveness=0.65,
                        duration_days=90
                    ),
                    BehavioralNudge(
                        nudge_type="loyalty_points",
                        target_segment=CustomerSegment.LOYAL_PREMIUM,
                        incentive_value=0.10,  # 10% points multiplier
                        psychological_trigger="status_and_exclusivity",
                        effectiveness=0.75,
                        duration_days=180
                    ),
                    BehavioralNudge(
                        nudge_type="convenience_bundles",
                        target_segment=CustomerSegment.CONVENIENCE_SEEKER,
                        incentive_value=0.12,  # 12% bundle discount
                        psychological_trigger="simplicity_and_time_saving",
                        effectiveness=0.58,
                        duration_days=60
                    ),
                    BehavioralNudge(
                        nudge_type="quality_assurance",
                        target_segment=CustomerSegment.QUALITY_FOCUSED,
                        incentive_value=0.08,  # 8% premium quality guarantee
                        psychological_trigger="trust_and_reliability",
                        effectiveness=0.52,
                        duration_days=120
                    ),
                    BehavioralNudge(
                        nudge_type="new_user_welcome",
                        target_segment=CustomerSegment.OCCASIONAL_USER,
                        incentive_value=0.20,  # 20% first order discount
                        psychological_trigger="fear_of_missing_out",
                        effectiveness=0.82,
                        duration_days=30
                    )
                ]
                
                return nudges
            
            def _create_partnership_programs(self) -> List[PartnershipProgram]:
                """Create strategic partnership programs"""
                
                programs = [
                    PartnershipProgram(
                        program_id="REST_001",
                        partner_type="restaurant",
                        partner_name="Local Eats Partnership",
                        benefits=["increased_order_volume", "marketing_exposure", "data_insights"],
                        commission_rate=0.08,  # 8% commission
                        customer_acquisition_rate=0.15,  # 15% new customers
                        retention_improvement=0.12  # 12% better retention
                    ),
                    PartnershipProgram(
                        program_id="CORP_001",
                        partner_type="corporate",
                        partner_name="Corporate Lunch Program",
                        benefits=["stable_revenue", "bulk_orders", "brand_loyalty"],
                        commission_rate=0.12,  # 12% commission
                        customer_acquisition_rate=0.25,  # 25% new customers
                        retention_improvement=0.20  # 20% better retention
                    ),
                    PartnershipProgram(
                        program_id="LOY_001",
                        partner_type="loyalty",
                        partner_name="Premium Rewards Network",
                        benefits=["cross_promotion", "premium_customers", "data_sharing"],
                        commission_rate=0.05,  # 5% commission
                        customer_acquisition_rate=0.18,  # 18% new customers
                        retention_improvement=0.25  # 25% better retention
                    ),
                ]
                
                return programs
            
            def simulate_demand_shaping(self, ecosystem: MarketEcosystem, months: int = 12) -> Dict:
                """Simulate demand shaping strategies over time"""
                
                print(f"\nStarting demand shaping simulation for {months} months...")
                
                results = {
                    'monthly_metrics': [],
                    'segment_performance': {},
                    'nudge_effectiveness': {},
                    'partnership_impact': {},
                    'market_evolution': {}
                }
                
                # Initial state
                active_customers = ecosystem.market_metrics['active_customers']
                market_share = ecosystem.market_share
                profit_margin = ecosystem.profit_margin
                
                for month in range(months):
                    print(f"\nMonth {month + 1}:")
                    
                    # Apply pricing strategy
                    pricing_impact = self._apply_pricing_strategy(ecosystem)
                    
                    # Apply behavioral nudges
                    nudge_impact = self._apply_behavioral_nudges(ecosystem, month)
                    
                    # Apply partnership programs
                    partnership_impact = self._apply_partnership_programs(ecosystem)
                    
                    # Calculate monthly metrics
                    monthly_metrics = self._calculate_monthly_metrics(
                        ecosystem, active_customers, pricing_impact, nudge_impact, partnership_impact
                    )
                    
                    # Update ecosystem state
                    active_customers = int(monthly_metrics['active_customers'])
                    market_share = monthly_metrics['market_share']
                    profit_margin = monthly_metrics['profit_margin']
                    
                    # Update ecosystem metrics
                    ecosystem.market_metrics.update(monthly_metrics)
                    ecosystem.market_share = market_share
                    ecosystem.profit_margin = profit_margin
                    ecosystem.competitive_position = min(0.8, market_share * 2)  # Simplified competitive position
                    
                    # Store results
                    results['monthly_metrics'].append(monthly_metrics)
                    
                    # Print progress
                    print(f"  Active customers: {active_customers:,}")
                    print(f"  Market share: {market_share:.1%}")
                    print(f"  Profit margin: {profit_margin:.1%}")
                    print(f"  Monthly orders: {monthly_metrics['monthly_orders']:,}")
                    print(f"  Avg order value: ${monthly_metrics['avg_order_value']:.2f}")
                    print(f"  Peak demand: {monthly_metrics['peak_hour_demand']:.1%}")
                    
                    if month == months - 1:
                        # Final analysis
                        results['segment_performance'] = self._analyze_segment_performance(ecosystem)
                        results['nudge_effectiveness'] = self._analyze_nudge_effectiveness(ecosystem)
                        results['partnership_impact'] = self._analyze_partnership_impact(ecosystem)
                        results['market_evolution'] = self._analyze_market_evolution(results['monthly_metrics'])
                
                return results
            
            def _apply_pricing_strategy(self, ecosystem: MarketEcosystem) -> Dict:
                """Apply dynamic pricing strategy impact"""
                
                pricing = ecosystem.pricing_strategy
                
                # Calculate pricing impact by segment
                segment_impact = {}
                
                for segment, profiles in ecosystem.customer_segments.items():
                    if not profiles:
                        continue
                        
                    # Average elasticity for segment
                    avg_elasticity = np.mean([p.price_elasticity for p in profiles])
                    segment_adjustment = pricing.segment_pricing[segment]
                    
                    # Demand change due to pricing
                    demand_change = avg_elasticity * segment_adjustment
                    
                    # Peak vs off-peak demand shift
                    peak_shift = 0.1 * segment_adjustment  # 10% of price effect shifts demand
                    
                    segment_impact[segment] = {
                        'demand_change': demand_change,
                        'peak_shift': peak_shift,
                        'revenue_impact': segment_adjustment * (1 + demand_change)
                        'avg_order_change': segment_adjustment * 0.5  # Orders change half as much as price
                    }
                
                return segment_impact
            
            def _apply_behavioral_nudges(self, ecosystem: MarketEcosystem, month: int) -> Dict:
                """Apply behavioral nudges impact"""
                
                nudge_impact = {}
                
                for nudge in ecosystem.behavioral_nudges:
                    # Check if nudge is still active
                    if month >= nudge.duration_days / 30:
                        continue
                    
                    target_profiles = ecosystem.customer_segments.get(nudge.target_segment, [])
                    if not target_profiles:
                        continue
                    
                    # Calculate nudge effectiveness
                    avg_susceptibility = np.mean([p.nudge_susceptibility for p in target_profiles])
                    actual_effectiveness = nudge.effectiveness * avg_susceptibility
                    
                    # Calculate impact metrics
                    behavior_change = actual_effectiveness * nudge.incentive_value
                    order_frequency_change = behavior_change * 0.3  # 30% change in frequency
                    retention_improvement = behavior_change * 0.2  # 20% retention improvement
                    off_peak_shift = behavior_change * 0.4 if 'off_peak' in nudge.nudge_type else 0
                    
                    nudge_impact[nudge.nudge_type] = {
                        'effectiveness': actual_effectiveness,
                        'behavior_change': behavior_change,
                        'order_frequency_change': order_frequency_change,
                        'retention_improvement': retention_improvement,
                        'off_peak_shift': off_peak_shift
                    }
                
                return nudge_impact
            
            def _apply_partnership_programs(self, ecosystem: MarketEcosystem) -> Dict:
                """Apply partnership program impact"""
                
                partnership_impact = {}
                
                for program in ecosystem.partnership_programs:
                    # Calculate cumulative impact
                    new_customers = ecosystem.market_metrics['total_customers'] * program.customer_acquisition_rate * 0.01
                    retention_boost = program.retention_improvement
                    revenue_impact = 1 - program.commission_rate  # Net revenue after commission
                    
                    partnership_impact[program.program_id] = {
                        'new_customers': new_customers,
                        'retention_boost': retention_boost,
                        'revenue_impact': revenue_impact,
                        'partner_type': program.partner_type
                    }
                
                return partnership_impact
            
            def _calculate_monthly_metrics(self, ecosystem: MarketEcosystem, active_customers: int,
                                       pricing_impact: Dict, nudge_impact: Dict, partnership_impact: Dict) -> Dict:
                """Calculate monthly performance metrics"""
                
                # Base metrics
                base_orders_per_customer = 8.5  # Average orders per customer per month
                base_avg_order_value = 32.0  # Base average order value
                base_peak_demand = 0.4  # 40% in peak hours
                
                # Apply pricing impacts
                pricing_orders_change = 0
                pricing_value_change = 0
                pricing_peak_shift = 0
                
                for segment, impact in pricing_impact.items():
                    segment_customers = len(ecosystem.customer_segments.get(segment, []))
                    segment_weight = segment_customers / self.market_size
                    
                    pricing_orders_change += impact['demand_change'] * segment_weight
                    pricing_value_change += impact['avg_order_change'] * segment_weight
                    pricing_peak_shift += impact['peak_shift'] * segment_weight
                
                # Apply nudge impacts
                nudge_frequency_change = 0
                nudge_retention_boost = 0
                nudge_off_peak_shift = 0
                
                for nudge_type, impact in nudge_impact.items():
                    nudge_frequency_change += impact['order_frequency_change'] * 0.2  # Weighted by nudge importance
                    nudge_retention_boost += impact['retention_improvement'] * 0.15
                    nudge_off_peak_shift += impact['off_peak_shift'] * 0.3
                
                # Apply partnership impacts
                partnership_customers = sum([impact['new_customers'] for impact in partnership_impact.values()])
                partnership_retention = np.mean([impact['retention_boost'] for impact in partnership_impact.values()]) if partnership_impact else 0
                partnership_revenue = np.mean([impact['revenue_impact'] for impact in partnership_impact.values()]) if partnership_impact else 1.0
                
                # Calculate final metrics
                monthly_orders = int(active_customers * base_orders_per_customer * (1 + pricing_orders_change + nudge_frequency_change))
                avg_order_value = base_avg_order_value * (1 + pricing_value_change) * partnership_revenue
                peak_hour_demand = base_peak_demand + pricing_peak_shift + nudge_off_peak_shift
                
                # Update active customers
                new_active_customers = active_customers + int(partnership_customers)
                retention_rate = 0.85 + nudge_retention_boost + partnership_retention
                final_active_customers = int(new_active_customers * retention_rate)
                
                # Calculate market share and profit margin
                market_share = final_active_customers / self.market_size
                profit_margin = 0.12 + (market_share - 0.15) * 0.3  # Margin improves with market share
                
                return {
                    'active_customers': final_active_customers,
                    'monthly_orders': monthly_orders,
                    'avg_order_value': avg_order_value,
                    'peak_hour_demand': peak_hour_demand,
                    'market_share': market_share,
                    'profit_margin': profit_margin,
                    'monthly_revenue': monthly_orders * avg_order_value,
                    'monthly_profit': monthly_orders * avg_order_value * profit_margin
                }
            
            def _analyze_segment_performance(self, ecosystem: MarketEcosystem) -> Dict:
                """Analyze performance by customer segment"""
                
                segment_analysis = {}
                
                for segment, profiles in ecosystem.customer_segments.items():
                    if not profiles:
                        continue
                    
                    segment_metrics = {
                        'segment_size': len(profiles),
                        'avg_elasticity': np.mean([p.price_elasticity for p in profiles]),
                        'avg_loyalty': np.mean([p.loyalty_score for p in profiles]),
                        'avg_frequency': np.mean([p.order_frequency for p in profiles]),
                        'avg_value': np.mean([p.avg_order_value for p in profiles]),
                        'peak_preference': np.mean([1 if hour in p.preferred_hours else 0 
                                                for p in profiles for hour in [11, 12, 18, 19, 20]]) / len(profiles)
                    }
                    
                    segment_analysis[segment.value] = segment_metrics
                
                return segment_analysis
            
            def _analyze_nudge_effectiveness(self, ecosystem: MarketEcosystem) -> Dict:
                """Analyze behavioral nudge effectiveness"""
                
                nudge_analysis = {}
                
                for nudge in ecosystem.behavioral_nudges:
                    target_profiles = ecosystem.customer_segments.get(nudge.target_segment, [])
                    if not target_profiles:
                        continue
                    
                    avg_susceptibility = np.mean([p.nudge_susceptibility for p in target_profiles])
                    actual_effectiveness = nudge.effectiveness * avg_susceptibility
                    
                    nudge_analysis[nudge.nudge_type] = {
                        'target_segment': nudge.target_segment.value,
                        'base_effectiveness': nudge.effectiveness,
                        'actual_effectiveness': actual_effectiveness,
                        'susceptibility_factor': avg_susceptibility,
                        'incentive_value': nudge.incentive_value,
                        'psychological_trigger': nudge.psychological_trigger
                    }
                
                return nudge_analysis
            
            def _analyze_partnership_impact(self, ecosystem: MarketEcosystem) -> Dict:
                """Analyze partnership program impact"""
                
                partnership_analysis = {}
                
                for program in ecosystem.partnership_programs:
                    partnership_analysis[program.program_id] = {
                        'partner_name': program.partner_name,
                        'partner_type': program.partner_type,
                        'commission_rate': program.commission_rate,
                        'acquisition_rate': program.customer_acquisition_rate,
                        'retention_improvement': program.retention_improvement,
                        'benefits': program.benefits
                    }
                
                return partnership_analysis
            
            def _analyze_market_evolution(self, monthly_metrics: List[Dict]) -> Dict:
                """Analyze market evolution over time"""
                
                if not monthly_metrics:
                    return {}
                
                # Calculate growth rates
                initial_customers = monthly_metrics[0]['active_customers']
                final_customers = monthly_metrics[-1]['active_customers']
                customer_growth = (final_customers - initial_customers) / initial_customers * 100
                
                initial_revenue = monthly_metrics[0]['monthly_revenue']
                final_revenue = monthly_metrics[-1]['monthly_revenue']
                revenue_growth = (final_revenue - initial_revenue) / initial_revenue * 100
                
                initial_margin = monthly_metrics[0]['profit_margin']
                final_margin = monthly_metrics[-1]['profit_margin']
                margin_improvement = (final_margin - initial_margin) * 100
                
                # Calculate peak demand shift
                initial_peak = monthly_metrics[0]['peak_hour_demand']
                final_peak = monthly_metrics[-1]['peak_hour_demand']
                peak_shift = (initial_peak - final_peak) / initial_peak * 100  # Reduction in peak demand
                
                return {
                    'customer_growth_rate': customer_growth,
                    'revenue_growth_rate': revenue_growth,
                    'margin_improvement': margin_improvement,
                    'peak_demand_reduction': peak_shift,
                    'initial_customers': initial_customers,
                    'final_customers': final_customers,
                    'initial_revenue': initial_revenue,
                    'final_revenue': final_revenue,
                    'months_analyzed': len(monthly_metrics)
                }
        
        # Create demand shaping optimizer
        optimizer = DemandShapingOptimizer()
        print("Demand shaping optimizer initialized")
    except Exception as e:
        print(f'Error in cell: {e}')
except Exception as e:
    print(f'Error in cell: {e}')
```
