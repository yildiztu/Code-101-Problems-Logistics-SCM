## 1. Physical-Cyber System Architecture

### Programmable Matter Components:

1. **Smart Containers**: Containers with adjustable internal dimensions and properties
2. **Shape-Shifting Items**: Cargo that can modify its shape for optimal packing
3. **Nano-Actuators**: Microscopic devices that enable physical transformation
4. **Distributed Intelligence**: Computational elements embedded throughout the system
5. **Energy Management**: Power systems for sustained physical operations

```python
try:
    @dataclass
    class ProgrammableMatterProperties:
        """Properties of programmable matter components"""
        min_density: float = 100.0  # kg/m³ (minimum achievable density)
        max_density: float = 2000.0  # kg/m³ (maximum achievable density)
        shape_change_rate: float = 0.1  # m/s (rate of shape transformation)
        energy_consumption: float = 50.0  # W/m³ (energy per unit volume)
        precision_level: float = 0.001  # m (minimum feature size)
        response_time: float = 0.5  # s (time to respond to commands)
        durability_cycles: int = 1000000  # number of shape-change cycles
    
    @dataclass
    class SmartItem:
        """Item with programmable matter capabilities"""
        id: int
        base_dimensions: Tuple[float, float, float]  # (length, width, height) in meters
        base_mass: float  # kg
        programmable: bool = True
        matter_properties: ProgrammableMatterProperties = field(default_factory=ProgrammableMatterProperties)
        current_shape: str = "rectangular"  # current shape configuration
        current_density: float = 500.0  # kg/m³
        deformation_capability: float = 0.3  # maximum deformation ratio (30%)
        
        def get_current_dimensions(self) -> Tuple[float, float, float]:
            """Get current dimensions based on shape configuration"""
            if self.current_shape == "rectangular":
                return self.base_dimensions
            elif self.current_shape == "compressed":
                factor = 0.7  # 30% compression
                return tuple(d * factor for d in self.base_dimensions)
            elif self.current_shape == "expanded":
                factor = 1.2  # 20% expansion
                return tuple(d * factor for d in self.base_dimensions)
            else:
                return self.base_dimensions
        
        def get_current_volume(self) -> float:
            """Calculate current volume based on shape"""
            dims = self.get_current_dimensions()
            return dims[0] * dims[1] * dims[2]
        
        def get_current_mass(self) -> float:
            """Calculate current mass based on density and volume"""
            return self.current_density * self.get_current_volume()
    
    @dataclass
    class SmartContainer:
        """Container with programmable matter capabilities"""
        id: int
        base_dimensions: Tuple[float, float, float]  # (length, width, height) in meters
        max_mass_capacity: float  # kg
        programmable: bool = True
        matter_properties: ProgrammableMatterProperties = field(default_factory=ProgrammableMatterProperties)
        current_internal_volume: float = 0.0  # m³
        compartment_config: str = "single"  # single, multi, adaptive
        wall_thickness: float = 0.1  # m
        
        def get_internal_dimensions(self) -> Tuple[float, float, float]:
            """Get internal usable dimensions"""
            return tuple(d - 2 * self.wall_thickness for d in self.base_dimensions)
        
        def get_internal_volume(self) -> float:
            """Calculate internal volume"""
            dims = self.get_internal_dimensions()
            return dims[0] * dims[1] * dims[2]
        
        def reconfigure_compartment(self, new_config: str) -> bool:
            """Reconfigure container compartments"""
            if self.programmable:
                self.compartment_config = new_config
                return True
            return False
    
    print("🔧 Physical-Cyber system architecture defined")
except Exception as e:
    print(f'Error in cell: {e}')
```

## 2. Programmable Matter Optimization Algorithm

### Core Algorithm: Self-Organizing Physical System

The algorithm leverages the unique capabilities of programmable matter to achieve optimal packing through:

1. **Dynamic Shape Optimization**: Items can change shape to fit available spaces
2. **Density Modulation**: Adjust material density to utilize vertical space efficiently
3. **Compartment Reconfiguration**: Container walls can move to create optimal spaces
4. **Real-time Adaptation**: Continuous optimization based on loading progress
5. **Distributed Coordination**: Local decision-making without central control

```python
try:
    class ProgrammableMatterOptimizer:
        """Advanced optimizer using programmable matter capabilities"""
        
        def __init__(self, container: SmartContainer, items: List[SmartItem]):
            self.container = container
            self.items = items
           .placed_items = []
           .unplaced_items = items.copy()
           .optimization_history = []
           .energy_consumed = 0.0
           .shape_changes = 0
            
        def analyze_space_requirements(self) -> Dict:
            """Analyze total space requirements and optimization potential"""
            total_base_volume = sum(item.get_current_volume() for item in self.items)
            container_volume = self.container.get_internal_volume()
            
            # Calculate compression potential
            compression_savings = 0
            for item in self.items:
                if item.programmable:
                    base_vol = item.get_current_volume()
                    compressed_vol = item.get_current_volume() * 0.7**3  # 30% compression in all dimensions
                    compression_savings += (base_vol - compressed_vol)
            
            return {
                'total_item_volume': total_base_volume,
                'container_volume': container_volume,
                'space_utilization': total_base_volume / container_volume,
                'compression_potential': compression_savings,
                'optimized_utilization': (total_base_volume - compression_savings) / container_volume
            }
        
        def optimize_item_shapes(self) -> List[SmartItem]:
            """Optimize item shapes for better packing"""
            optimized_items = []
            
            for item in self.unplaced_items:
                if item.programmable:
                    # Determine optimal shape based on remaining space
                    remaining_volume = self.container.get_internal_volume() - sum(
                        placed.get_current_volume() for placed in self.placed_items
                    )
                    
                    # Simple heuristic: compress if space is tight
                    if remaining_volume < item.get_current_volume() * 1.5:
                        item.current_shape = "compressed"
                        self.shape_changes += 1
                        self.energy_consumed += item.matter_properties.energy_consumption * item.get_current_volume()
                    
                optimized_items.append(item)
            
            return optimized_items
        
        def find_optimal_position(self, item: SmartItem) -> Optional[Tuple[float, float, float]]:
            """Find optimal position for item using programmable matter capabilities"""
            item_dims = item.get_current_dimensions()
            container_dims = self.container.get_internal_dimensions()
            
            # Try different positions with fine granularity
            step = 0.05  # 5cm granularity
            best_position = None
            best_fitness = -float('inf')
            
            for x in np.arange(0, container_dims[0] - item_dims[0] + step, step):
                for y in np.arange(0, container_dims[1] - item_dims[1] + step, step):
                    for z in np.arange(0, container_dims[2] - item_dims[2] + step, step):
                        position = (x, y, z)
                        
                        if self._is_position_valid(position, item_dims):
                            # Calculate fitness (prefer lower center of gravity and corner placement)
                            fitness = self._calculate_position_fitness(position, item_dims)
                            
                            if fitness > best_fitness:
                                best_fitness = fitness
                                best_position = position
            
            return best_position
        
        def _is_position_valid(self, position: Tuple[float, float, float], 
                              item_dims: Tuple[float, float, float]) -> bool:
            """Check if position is valid (no collisions)"""
            x, y, z = position
            
            # Check container boundaries
            container_dims = self.container.get_internal_dimensions()
            if (x < 0 or y < 0 or z < 0 or 
                x + item_dims[0] > container_dims[0] or 
                y + item_dims[1] > container_dims[1] or 
                z + item_dims[2] > container_dims[2]):
                return False
            
            # Check collisions with placed items
            for placed_item, placed_pos in self.placed_items:
                placed_dims = placed_item.get_current_dimensions()
                px, py, pz = placed_pos
                
                # Simple AABB collision detection
                if not (x + item_dims[0] <= px or px + placed_dims[0] <= x or
                        y + item_dims[1] <= py or py + placed_dims[1] <= y or
                        z + item_dims[2] <= pz or pz + placed_dims[2] <= z):
                    return False
            
            return True
        
        def _calculate_position_fitness(self, position: Tuple[float, float, float], 
                                      item_dims: Tuple[float, float, float]) -> float:
            """Calculate fitness score for position"""
            x, y, z = position
            
            # Prefer lower center of gravity (minimize z)
            gravity_fitness = 1.0 / (1.0 + z)
            
            # Prefer corner placement (minimize distance from origin)
            corner_fitness = 1.0 / (1.0 + x + y)
            
            # Prefer stable placement (larger base area)
            stability_fitness = (item_dims[0] * item_dims[1]) / max(item_dims)
            
            return gravity_fitness * 0.4 + corner_fitness * 0.3 + stability_fitness * 0.3
        
        def pack_items(self) -> Dict:
            """Main packing algorithm using programmable matter"""
            start_time = time.time()
            
            # Analyze initial space requirements
            space_analysis = self.analyze_space_requirements()
            
            # Optimize item shapes
            optimized_items = self.optimize_item_shapes()
            
            # Sort items by volume (largest first for better space utilization)
            optimized_items.sort(key=lambda x: x.get_current_volume(), reverse=True)
            
            # Pack items
            for item in optimized_items:
                position = self.find_optimal_position(item)
                
                if position:
                    self.placed_items.append((item, position))
                    self.unplaced_items.remove(item)
                    
                    # Record optimization step
                    self.optimization_history.append({
                        'item_id': item.id,
                        'position': position,
                        'shape': item.current_shape,
                        'volume': item.get_current_volume(),
                        'placed_count': len(self.placed_items)
                    })
            
            end_time = time.time()
            
            # Calculate final metrics
            total_placed_volume = sum(item.get_current_volume() for item, _ in self.placed_items)
            container_volume = self.container.get_internal_volume()
            
            return {
                'placed_items': len(self.placed_items),
                'unplaced_items': len(self.unplaced_items),
                'space_utilization': total_placed_volume / container_volume,
                'total_volume_placed': total_placed_volume,
                'shape_changes': self.shape_changes,
                'energy_consumed': self.energy_consumed,
                'computation_time': end_time - start_time,
                'space_analysis': space_analysis
            }
    
    print("🚀 Programmable Matter Optimization Algorithm implemented")
except Exception as e:
    print(f'Error in cell: {e}')
```
