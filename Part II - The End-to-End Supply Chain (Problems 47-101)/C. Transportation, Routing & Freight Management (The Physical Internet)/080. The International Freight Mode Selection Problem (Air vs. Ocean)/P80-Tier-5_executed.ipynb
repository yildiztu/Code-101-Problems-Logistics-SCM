# 80. The International Freight Mode Selection Problem (Air vs. Ocean)
## Tier 5: The Integrated Digital Twin (System-of-Systems Simulation)

### Key assumptions
- Digital twin integrates multiple logistics subsystems in real-time
- IoT sensors provide continuous data feeds from physical operations
- Predictive analytics enable proactive decision-making and scenario planning
- System-of-systems approach captures complex interdependencies
- Real-time synchronization allows dynamic mode selection optimization

### Approach (step-by-step)
1. **Model physical assets**: Digital replicas of ships, ports, and cargo
2. **Establish connectivity**: IoT sensor networks and data pipelines
3. **Implement data processing**: Real-time analytics and pattern recognition
4. **Create application layer**: Decision support and optimization engines
5. **Enable simulation**: Monte Carlo scenario testing and what-if analysis
6. **Integrate optimization**: Dynamic freight mode selection with live data
7. **Monitor KPIs**: Performance metrics and system health indicators

### What to look for in the results
- Real-time data integration from multiple logistics systems
- Predictive analytics for market disruption forecasting
- Scenario analysis showing system resilience and adaptation
- KPI dashboards with comprehensive performance monitoring
- Dynamic optimization responding to live market conditions

### Concrete example (from the source)
TechGlobal's Digital Twin processes 2.3 million data points daily from 47 integrated systems:
- **Physical Assets**: 200+ vessels, 15+ ports, 5000+ containers tracked
- **Connectivity Layer**: 500 IoT sensors, 50 RFID readers, satellite tracking
- **Data Processing**: Real-time analytics with 1-minute update cycles
- **Application Layer: AI-optimized coordination and predictive alerts

During Hurricane Season 2024, the twin predicted 73% probability of 5+ day delays for Asia-US West Coast ocean routes. The system automatically recommended shifting 40% of time-sensitive cargo to air freight three days before actual disruptions occurred, preventing $2.8M in late delivery penalties while competitors experienced average delays of 8.2 days.

```python
# Import required packages for digital twin implementation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import time
from datetime import datetime, timedelta
from dataclasses import dataclass
from typing import List, Dict, Tuple, Optional
import warnings
warnings.filterwarnings('ignore')

# Set up visualization style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
@dataclass
class Shipment:
    """Data class for shipment information"""
    id: int
    weight: float  # kg
    value: float   # USD
    destination: str
    due_date: int  # days from now
    priority: str  # high, medium, low

@dataclass 
class TransportParams:
    """Parameters for transportation costs and times"""
    costs: Dict[str, Dict[str, float]]  # mode -> destination -> cost per kg
    transit_times: Dict[str, Dict[str, int]]  # mode -> destination -> days
    holding_rate: float  # daily holding cost rate as fraction of value
    late_penalty: float  # penalty per day late
    air_capacity: float  # monthly air freight capacity in kg

@dataclass
class IoTSensor:
    """IoT sensor data for real-time monitoring"""
    sensor_id: str
    sensor_type: str  # position, temperature, humidity, weight, etc.
    location: str
    timestamp: datetime
    value: float
    unit: str

@dataclass
class MarketData:
    """Real-time market data feeds"""
    timestamp: datetime
    ocean_rate_index: float
    air_rate_index: float
    fuel_price_index: float
    port_congestion_index: Dict[str, float]
    weather_disruption_risk: Dict[str, float]

# Create comprehensive digital twin environment
def create_digital_twin_environment() -> Tuple[List[Shipment], TransportParams, Dict]:
    """
    Create a comprehensive digital twin environment with multiple data sources
    """
    # Extended shipment set for digital twin
    shipments = [
        Shipment(1, 800, 1500000, "rotterdam", 8, "high"),    # Critical, time-sensitive
        Shipment(2, 1200, 800000, "los_angeles", 15, "medium"), # Standard priority
        Shipment(3, 3500, 600000, "rotterdam", 25, "low"),    # Low priority, heavy
        Shipment(4, 600, 2000000, "singapore", 10, "high"),   # High value, urgent
        Shipment(5, 2000, 400000, "hamburg", 20, "medium"),   # Medium priority
        Shipment(6, 1500, 900000, "new_york", 12, "high"),   # High priority
        Shipment(7, 4000, 300000, "rotterdam", 35, "low"),    # Bulk, low urgency
        Shipment(8, 900, 1200000, "los_angeles", 18, "medium") # Medium-high value
    ]
    
    # Base transportation parameters
    base_params = TransportParams(
        costs={
            "ocean": {
                "rotterdam": 2.2, "los_angeles": 1.9, "singapore": 2.8,
                "hamburg": 2.1, "new_york": 1.8
            },
            "air": {
                "rotterdam": 6.8, "los_angeles": 6.2, "singapore": 7.5,
                "hamburg": 6.5, "new_york": 5.9
            }
        },
        transit_times={
            "ocean": {
                "rotterdam": 28, "los_angeles": 20, "singapore": 35,
                "hamburg": 26, "new_york": 18
            },
            "air": {
                "rotterdam": 3, "los_angeles": 2, "singapore": 4,
                "hamburg": 3, "new_york": 2
            }
        },
        holding_rate=0.0012,
        late_penalty=1500,
        air_capacity=15000
    )
    
    # Digital twin components
    digital_twin_components = {
        'physical_assets': {
            'vessels': 25,  # Number of vessels in fleet
            'ports': 8,     # Number of ports served
            'containers': 5000,  # Container fleet size
            'warehouses': 12  # Distribution centers
        },
        'connectivity': {
            'iot_sensors': 500,
            'rfid_readers': 50,
            'satellite_links': 15,
            'data_sources': 47  # Number of integrated systems
        },
        'data_processing': {
            'update_frequency': 60,  # seconds
            'data_points_per_day': 2300000,
            'analytics_latency': 30,  # seconds
            'prediction_horizon': 7  # days
        }
    }
    
    return shipments, base_params, digital_twin_components

# Initialize digital twin environment
shipments, base_params, twin_components = create_digital_twin_environment()

print("Digital Twin Environment Initialized:")
print(f"  Shipments: {len(shipments)}")
print(f"  Physical Assets: {twin_components['physical_assets']}")
print(f"  Connectivity: {twin_components['connectivity']}")
print(f"  Data Processing: {twin_components['data_processing']}")
```

```python
class IoTDataSimulator:
    """
    Simulate IoT sensor data for digital twin integration
    Based on the IoT integration from the source material
    """
    
    def __init__(self, twin_components: Dict):
        self.twin_components = twin_components
        self.sensor_types = ['position', 'temperature', 'humidity', 'weight', 'vibration', 'pressure']
        self.locations = ['rotterdam', 'los_angeles', 'singapore', 'hamburg', 'new_york',
                         'vessel_1', 'vessel_2', 'warehouse_1', 'port_rotterdam', 'port_la']
        
        # Initialize sensor network
        self.sensors = self._initialize_sensors()
        self.current_time = datetime.now()
    
    def _initialize_sensors(self) -> List[IoTSensor]:
        """
        Initialize IoT sensor network
        """
        sensors = []
        sensor_id = 0
        
        # Create sensors for each location and type
        for location in self.locations:
            for sensor_type in self.sensor_types:
                # Distribute sensors across locations
                num_sensors = max(1, self.twin_components['connectivity']['iot_sensors'] // 
                                (len(self.locations) * len(self.sensor_types)))
                
                for i in range(num_sensors):
                    sensors.append(IoTSensor(
                        sensor_id=f"{location}_{sensor_type}_{i}",
                        sensor_type=sensor_type,
                        location=location,
                        timestamp=self.current_time,
                        value=self._generate_sensor_value(sensor_type),
                        unit=self._get_sensor_unit(sensor_type)
                    ))
                    sensor_id += 1
        
        return sensors
    
    def _generate_sensor_value(self, sensor_type: str) -> float:
        """
        Generate realistic sensor values based on type
        """
        if sensor_type == 'position':
            return random.uniform(-180, 180)  # Longitude
        elif sensor_type == 'temperature':
            return random.uniform(15, 35)     # Celsius
        elif sensor_type == 'humidity':
            return random.uniform(30, 80)     # Percentage
        elif sensor_type == 'weight':
            return random.uniform(100, 50000)  # kg
        elif sensor_type == 'vibration':
            return random.uniform(0, 10)       # Hz
        elif sensor_type == 'pressure':
            return random.uniform(980, 1050)   # hPa
        else:
            return random.uniform(0, 100)
    
    def _get_sensor_unit(self, sensor_type: str) -> str:
        """
        Get unit for sensor type
        """
        units = {
            'position': 'deg',
            'temperature': '°C',
            'humidity': '%',
            'weight': 'kg',
            'vibration': 'Hz',
            'pressure': 'hPa'
        }
        return units.get(sensor_type, 'unit')
    
    def generate_real_time_data(self, num_readings: int = 100) -> List[IoTSensor]:
        """
        Generate real-time sensor data readings
        """
        readings = []
        
        for _ in range(num_readings):
            # Select random sensor
            sensor = random.choice(self.sensors)
            
            # Generate new reading with some variation
            base_value = sensor.value
            variation = random.gauss(0, abs(base_value) * 0.05)  # 5% variation
            new_value = max(0, base_value + variation)
            
            # Create new reading
            reading = IoTSensor(
                sensor_id=sensor.sensor_id,
                sensor_type=sensor.sensor_type,
                location=sensor.location,
                timestamp=self.current_time,
                value=new_value,
                unit=sensor.unit
            )
            
            readings.append(reading)
            
            # Advance time slightly
            self.current_time += timedelta(seconds=random.uniform(1, 10))
        
        return readings

class MarketDataFeed:
    """
    Simulate real-time market data feeds
    """
    
    def __init__(self):
        self.current_time = datetime.now()
        self.destinations = ['rotterdam', 'los_angeles', 'singapore', 'hamburg', 'new_york']
        
        # Initialize market conditions
        self.base_conditions = {
            'ocean_rate_index': 1.0,
            'air_rate_index': 1.0,
            'fuel_price_index': 1.0,
            'port_congestion': {dest: 0.5 for dest in self.destinations},
            'weather_risk': {dest: 0.1 for dest in self.destinations}
        }
    
    def generate_market_data(self) -> MarketData:
        """
        Generate current market data with realistic variations
        """
        # Add market dynamics
        ocean_trend = random.gauss(0, 0.02)  # 2% daily volatility
        air_trend = random.gauss(0, 0.015)
        fuel_trend = random.gauss(0, 0.025)
        
        # Update indices
        ocean_rate = max(0.5, self.base_conditions['ocean_rate_index'] * (1 + ocean_trend))
        air_rate = max(0.5, self.base_conditions['air_rate_index'] * (1 + air_trend))
        fuel_price = max(0.5, self.base_conditions['fuel_price_index'] * (1 + fuel_trend))
        
        # Update port congestion
        port_congestion = {}
        for dest in self.destinations:
            congestion_change = random.gauss(0, 0.05)
            current_congestion = self.base_conditions['port_congestion'][dest]
            port_congestion[dest] = max(0, min(1, current_congestion + congestion_change))
        
        # Update weather risk
        weather_risk = {}
        for dest in self.destinations:
            risk_change = random.gauss(0, 0.03)
            current_risk = self.base_conditions['weather_risk'][dest]
            weather_risk[dest] = max(0, min(1, current_risk + risk_change))
        
        # Create market data object
        market_data = MarketData(
            timestamp=self.current_time,
            ocean_rate_index=ocean_rate,
            air_rate_index=air_rate,
            fuel_price_index=fuel_price,
            port_congestion_index=port_congestion,
            weather_disruption_risk=weather_risk
        )
        
        # Update base conditions for next iteration
        self.base_conditions.update({
            'ocean_rate_index': ocean_rate,
            'air_rate_index': air_rate,
            'fuel_price_index': fuel_price,
            'port_congestion': port_congestion,
            'weather_risk': weather_risk
        })
        
        self.current_time += timedelta(minutes=random.uniform(5, 15))
        
        return market_data

# Initialize data simulators
iot_simulator = IoTDataSimulator(twin_components)
market_feed = MarketDataFeed()

print("Data Simulators Initialized:")
print(f"  IoT Sensors: {len(iot_simulator.sensors)}")
print(f"  Market Destinations: {len(market_feed.destinations)}")
print(f"  Data Generation: Real-time simulation ready")
```

```python
class DigitalTwinAnalytics:
    """
    Digital twin analytics engine for real-time processing and prediction
    Based on the analytics capabilities from the source material
    """
    
    def __init__(self, shipments: List[Shipment], params: TransportParams):
        self.shipments = shipments
        self.params = params
        
        # Analytics components
        self.historical_data = {
            'sensor_readings': [],
            'market_data': [],
            'decisions': [],
            'outcomes': []
        }
        
        # Prediction models
        self.disruption_predictions = {}
        self.cost_forecasts = {}
        self.capacity_alerts = []
        
        # KPI tracking
        self.kpis = {
            'on_time_delivery': [],
            'cost_efficiency': [],
            'capacity_utilization': [],
            'disruption_response_time': [],
            'prediction_accuracy': []
        }
    
    def process_sensor_data(self, sensor_readings: List[IoTSensor]) -> Dict:
        """
        Process real-time sensor data and extract insights
        """
        # Store readings
        self.historical_data['sensor_readings'].extend(sensor_readings)
        
        # Analyze sensor patterns
        analysis = {
            'total_readings': len(sensor_readings),
            'readings_by_type': {},
            'readings_by_location': {},
            'anomalies': [],
            'alerts': []
        }
        
        # Categorize readings
        for reading in sensor_readings:
            # By type
            if reading.sensor_type not in analysis['readings_by_type']:
                analysis['readings_by_type'][reading.sensor_type] = []
            analysis['readings_by_type'][reading.sensor_type].append(reading.value)
            
            # By location
            if reading.location not in analysis['readings_by_location']:
                analysis['readings_by_location'][reading.location] = []
            analysis['readings_by_location'][reading.location].append(reading.value)
            
            # Anomaly detection
            if self._detect_anomaly(reading):
                analysis['anomalies'].append({
                    'sensor_id': reading.sensor_id,
                    'type': reading.sensor_type,
                    'location': reading.location,
                    'value': reading.value,
                    'severity': 'high' if abs(reading.value) > 1000 else 'medium'
                })
        
        # Generate alerts
        analysis['alerts'] = self._generate_alerts(analysis['anomalies'])
        
        return analysis
    
    def _detect_anomaly(self, reading: IoTSensor) -> bool:
        """
        Simple anomaly detection based on value thresholds
        """
        # Define anomaly thresholds for different sensor types
        thresholds = {
            'temperature': (-10, 50),      # Outside normal range
            'humidity': (0, 100),          # Outside possible range
            'vibration': (0, 20),           # Excessive vibration
            'pressure': (900, 1100),       # Extreme pressure
            'weight': (0, 100000),         # Unreasonable weight
        }
        
        if reading.sensor_type in thresholds:
            min_val, max_val = thresholds[reading.sensor_type]
            return reading.value < min_val or reading.value > max_val
        
        return False
    
    def _generate_alerts(self, anomalies: List[Dict]) -> List[Dict]:
        """
        Generate alerts based on detected anomalies
        """
        alerts = []
        
        for anomaly in anomalies:
            if anomaly['severity'] == 'high':
                alerts.append({
                    'type': 'critical_anomaly',
                    'message': f"Critical anomaly detected in {anomaly['type']} sensor at {anomaly['location']}",
                    'sensor_id': anomaly['sensor_id'],
                    'timestamp': datetime.now(),
                    'action_required': 'immediate_investigation'
                })
            else:
                alerts.append({
                    'type': 'warning',
                    'message': f"Warning: Anomalous reading in {anomaly['type']} sensor at {anomaly['location']}",
                    'sensor_id': anomaly['sensor_id'],
                    'timestamp': datetime.now(),
                    'action_required': 'monitor'
                })
        
        return alerts
    
    def process_market_data(self, market_data: MarketData) -> Dict:
        """
        Process market data and generate predictions
        """
        # Store market data
        self.historical_data['market_data'].append(market_data)
        
        # Analyze market conditions
        analysis = {
            'current_conditions': {
                'ocean_rate_trend': market_data.ocean_rate_index,
                'air_rate_trend': market_data.air_rate_index,
                'fuel_price_trend': market_data.fuel_price_index,
                'avg_congestion': np.mean(list(market_data.port_congestion_index.values())),
                'avg_weather_risk': np.mean(list(market_data.weather_disruption_risk.values()))
            },
            'predictions': self._generate_disruption_predictions(market_data),
            'recommendations': self._generate_market_recommendations(market_data)
        }
        
        return analysis
    
    def _generate_disruption_predictions(self, market_data: MarketData) -> Dict:
        """
        Generate disruption predictions based on market data
        """
        predictions = {}
        
        for destination in market_data.port_congestion_index:
            congestion = market_data.port_congestion_index[destination]
            weather_risk = market_data.weather_disruption_risk[destination]
            
            # Calculate disruption probability
            disruption_prob = min(0.95, congestion * 0.6 + weather_risk * 0.4)
            
            # Predict delay if disruption occurs
            if disruption_prob > 0.7:
                predicted_delay = random.randint(3, 10)  # 3-10 days
            elif disruption_prob > 0.4:
                predicted_delay = random.randint(1, 5)   # 1-5 days
            else:
                predicted_delay = 0
            
            predictions[destination] = {
                'disruption_probability': disruption_prob,
                'predicted_delay_days': predicted_delay,
                'confidence': min(0.9, disruption_prob * 1.2),  # Higher confidence for higher probabilities
                'recommended_action': self._get_recommended_action(disruption_prob, predicted_delay)
            }
        
        return predictions
    
    def _get_recommended_action(self, disruption_prob: float, predicted_delay: int) -> str:
        """
        Get recommended action based on disruption prediction
        """
        if disruption_prob > 0.8 and predicted_delay > 5:
            return 'immediate_air_diversion'
        elif disruption_prob > 0.6 and predicted_delay > 3:
            return 'partial_air_diversion'
        elif disruption_prob > 0.4:
            return 'increase_buffer_inventory'
        else:
            return 'monitor_closely'
    
    def _generate_market_recommendations(self, market_data: MarketData) -> List[str]:
        """
        Generate market-based recommendations
        """
        recommendations = []
        
        # Ocean rate recommendations
        if market_data.ocean_rate_index > 1.3:
            recommendations.append("Ocean rates significantly elevated - consider air freight for time-sensitive cargo")
        elif market_data.ocean_rate_index < 0.8:
            recommendations.append("Ocean rates favorable - maximize ocean freight utilization")
        
        # Air rate recommendations
        if market_data.air_rate_index > 1.2:
            recommendations.append("Air rates elevated - restrict to critical shipments only")
        
        # Fuel price recommendations
        if market_data.fuel_price_index > 1.25:
            recommendations.append("Fuel prices high - optimize routes and consider fuel surcharges")
        
        # Capacity recommendations
        avg_congestion = np.mean(list(market_data.port_congestion_index.values()))
        if avg_congestion > 0.7:
            recommendations.append("Port congestion high - book capacity in advance and consider alternative ports")
        
        return recommendations
    
    def optimize_freight_modes(self, market_data: MarketData) -> Dict:
        """
        Optimize freight modes based on current market conditions and predictions
        """
        # Get disruption predictions
        predictions = self._generate_disruption_predictions(market_data)
        
        # Optimize each shipment
        optimization_results = {
            'assignments': {},
            'total_cost': 0,
            'risk_mitigation': {},
            'alternative_plans': {}
        }
        
        # Adjust parameters based on market conditions
        adjusted_params = self._adjust_parameters_for_market(market_data)
        
        for shipment in self.shipments:
            # Get disruption risk for destination
            dest_risk = predictions.get(shipment.destination, 
                                        {'disruption_probability': 0.1, 'predicted_delay_days': 0})
            
            # Calculate costs for both modes with risk adjustment
            ocean_cost = self._calculate_adjusted_cost(shipment, 'ocean', adjusted_params, dest_risk)
            air_cost = self._calculate_adjusted_cost(shipment, 'air', adjusted_params, dest_risk)
            
            # Make decision based on priority and risk
            if shipment.priority == 'high' and dest_risk['disruption_probability'] > 0.5:
                assigned_mode = 'air'  # High priority cargo gets air freight under risk
            elif ocean_cost < air_cost * 0.8:  # Significant cost advantage
                assigned_mode = 'ocean'
            elif dest_risk['predicted_delay_days'] > 3 and shipment.priority in ['high', 'medium']:
                assigned_mode = 'air'
            else:
                assigned_mode = 'ocean'
            
            # Calculate final cost
            final_cost = self._calculate_adjusted_cost(shipment, assigned_mode, adjusted_params, dest_risk)
            
            optimization_results['assignments'][shipment.id] = {
                'mode': assigned_mode,
                'cost': final_cost,
                'risk_adjusted_cost': final_cost,
                'disruption_probability': dest_risk['disruption_probability'],
                'recommended_action': dest_risk['recommended_action']
            }
            
            optimization_results['total_cost'] += final_cost
        
        return optimization_results
    
    def _adjust_parameters_for_market(self, market_data: MarketData) -> TransportParams:
        """
        Adjust transportation parameters based on current market conditions
        """
        adjusted_costs = {}
        for mode in ['ocean', 'air']:
            adjusted_costs[mode] = {}
            multiplier = market_data.ocean_rate_index if mode == 'ocean' else market_data.air_rate_index
            for dest in self.params.costs[mode]:
                # Add congestion surcharge
                congestion_surcharge = market_data.port_congestion_index[dest] * 0.2
                adjusted_costs[mode][dest] = self.params.costs[mode][dest] * multiplier * (1 + congestion_surcharge)
        
        # Adjust transit times for congestion
        adjusted_transit_times = {}
        for mode in ['ocean', 'air']:
            adjusted_transit_times[mode] = {}
            for dest in self.params.transit_times[mode]:
                congestion_delay = int(market_data.port_congestion_index[dest] * 3)  # Up to 3 days delay
                adjusted_transit_times[mode][dest] = self.params.transit_times[mode][dest] + congestion_delay
        
        # Adjust air capacity for congestion
        capacity_reduction = 1 - np.mean(list(market_data.port_congestion_index.values())) * 0.3
        adjusted_air_capacity = self.params.air_capacity * capacity_reduction
        
        return TransportParams(
            costs=adjusted_costs,
            transit_times=adjusted_transit_times,
            holding_rate=self.params.holding_rate,
            late_penalty=self.params.late_penalty,
            air_capacity=adjusted_air_capacity
        )
    
    def _calculate_adjusted_cost(self, shipment: Shipment, mode: str, 
                               params: TransportParams, risk: Dict) -> float:
        """
        Calculate cost with risk adjustment
        """
        # Base transportation cost
        transport_cost = shipment.weight * params.costs[mode][shipment.destination]
        
        # Transit time
        transit_time = params.transit_times[mode][shipment.destination]
        
        # Add expected delay cost
        expected_delay_cost = risk['predicted_delay_days'] * params.late_penalty * risk['disruption_probability']
        
        # Holding cost
        if transit_time < shipment.due_date:
            holding_cost = (shipment.due_date - transit_time) * params.holding_rate * shipment.value
        else:
            holding_cost = 0
        
        # Late penalty
        if transit_time > shipment.due_date:
            late_days = transit_time - shipment.due_date
            late_penalty_cost = late_days * params.late_penalty
        else:
            late_penalty_cost = 0
        
        return transport_cost + holding_cost + late_penalty_cost + expected_delay_cost

# Initialize digital twin analytics
analytics = DigitalTwinAnalytics(shipments, base_params)
print("Digital Twin Analytics initialized")
print(f"  Analytics components: Sensor processing, Market analysis, Predictive modeling")
print(f"  KPI tracking: 5 key performance indicators")
```

```python
class DigitalTwinSimulation:
    """
    Main digital twin simulation engine
    Integrates all components for real-time freight mode optimization
    """
    
    def __init__(self, shipments: List[Shipment], params: TransportParams, 
                 twin_components: Dict):
        self.shipments = shipments
        self.params = params
        self.twin_components = twin_components
        
        # Initialize components
        self.iot_simulator = IoTDataSimulator(twin_components)
        self.market_feed = MarketDataFeed()
        self.analytics = DigitalTwinAnalytics(shipments, params)
        
        # Simulation state
        self.current_time = datetime.now()
        self.simulation_step = 0
        
        # Results tracking
        self.simulation_results = {
            'sensor_analytics': [],
            'market_analytics': [],
            'optimization_results': [],
            'alerts': [],
            'kpis': [],
            'scenario_analysis': {}
        }
    
    def run_simulation_step(self) -> Dict:
        """
        Run one step of the digital twin simulation
        """
        step_results = {
            'timestamp': self.current_time,
            'step': self.simulation_step,
            'sensor_data': None,
            'market_data': None,
            'optimization': None,
            'alerts': [],
            'kpis': {}
        }
        
        # 1. Generate sensor data
        sensor_readings = self.iot_simulator.generate_real_time_data(50)
        sensor_analysis = self.analytics.process_sensor_data(sensor_readings)
        step_results['sensor_data'] = sensor_analysis
        
        # 2. Generate market data
        market_data = self.market_feed.generate_market_data()
        market_analysis = self.analytics.process_market_data(market_data)
        step_results['market_data'] = market_analysis
        
        # 3. Optimize freight modes
        optimization_results = self.analytics.optimize_freight_modes(market_data)
        step_results['optimization'] = optimization_results
        
        # 4. Collect alerts
        step_results['alerts'] = sensor_analysis.get('alerts', [])
        
        # 5. Calculate KPIs
        step_results['kpis'] = self._calculate_step_kpis(step_results)
        
        # Store results
        self.simulation_results['sensor_analytics'].append(sensor_analysis)
        self.simulation_results['market_analytics'].append(market_analysis)
        self.simulation_results['optimization_results'].append(optimization_results)
        self.simulation_results['alerts'].extend(step_results['alerts'])
        self.simulation_results['kpis'].append(step_results['kpis'])
        
        # Advance simulation
        self.current_time += timedelta(minutes=30)  # 30-minute steps
        self.simulation_step += 1
        
        return step_results
    
    def _calculate_step_kpis(self, step_results: Dict) -> Dict:
        """
        Calculate KPIs for current simulation step
        """
        kpis = {}
        
        # Cost efficiency
        if step_results['optimization']:
            total_cost = step_results['optimization']['total_cost']
            total_value = sum(s.value for s in self.shipments)
            kpis['cost_efficiency'] = total_cost / total_value if total_value > 0 else 0
            
            # Air capacity utilization
            air_assignments = sum(1 for assignment in step_results['optimization']['assignments'].values() 
                               if assignment['mode'] == 'air')
            air_weight = sum(s.weight for s in self.shipments 
                           if step_results['optimization']['assignments'][s.id]['mode'] == 'air')
            
            # Get current air capacity (adjusted for market conditions)
            market_data = step_results['market_data']['current_conditions']
            capacity_factor = 1 - market_data['avg_congestion'] * 0.3
            current_capacity = self.params.air_capacity * capacity_factor
            
            kpis['air_capacity_utilization'] = air_weight / current_capacity if current_capacity > 0 else 0
            kpis['air_shipment_percentage'] = air_assignments / len(self.shipments)
            
            # Risk exposure
            total_risk = sum(assignment['disruption_probability'] 
                           for assignment in step_results['optimization']['assignments'].values())
            kpis['average_risk_exposure'] = total_risk / len(self.shipments)
        
        # Data quality KPIs
        if step_results['sensor_data']:
            kpis['sensor_data_volume'] = step_results['sensor_data']['total_readings']
            kpis['anomaly_rate'] = len(step_results['sensor_data']['anomalies']) / step_results['sensor_data']['total_readings']
        
        # Alert KPIs
        kpis['critical_alerts'] = len([alert for alert in step_results['alerts'] if alert['type'] == 'critical_anomaly'])
        kpis['warning_alerts'] = len([alert for alert in step_results['alerts'] if alert['type'] == 'warning'])
        
        return kpis
    
    def run_scenario_analysis(self, scenario_name: str, num_steps: int = 20) -> Dict:
        """
        Run scenario analysis for specific market conditions
        """
        print(f"\nRunning {scenario_name} scenario analysis...")
        
        # Configure scenario
        if scenario_name == 'hurricane_season':
            # Simulate hurricane conditions
            self.market_feed.base_conditions.update({
                'ocean_rate_index': 1.4,
                'air_rate_index': 1.2,
                'weather_risk': {dest: 0.8 for dest in self.market_feed.destinations}
            })
        elif scenario_name == 'fuel_crisis':
            # Simulate fuel price spike
            self.market_feed.base_conditions.update({
                'fuel_price_index': 1.8,
                'ocean_rate_index': 1.3,
                'air_rate_index': 1.6
            })
        elif scenario_name == 'port_congestion':
            # Simulate severe port congestion
            self.market_feed.base_conditions.update({
                'port_congestion': {dest: 0.9 for dest in self.market_feed.destinations}
                'air_rate_index': 1.3
            })
        
        # Run simulation
        scenario_results = []
        for step in range(num_steps):
            step_result = self.run_simulation_step()
            scenario_results.append(step_result)
            
            if step % 5 == 0:
                print(f"  Step {step}: Cost=${step_result['optimization']['total_cost']:,.0f}, "
                      f"Alerts={len(step_result['alerts'])}")
        
        # Analyze scenario results
        scenario_analysis = {
            'scenario_name': scenario_name,
            'total_steps': num_steps,
            'final_cost': scenario_results[-1]['optimization']['total_cost'],
            'cost_evolution': [step['optimization']['total_cost'] for step in scenario_results],
            'alert_frequency': len(self.simulation_results['alerts']) / num_steps,
            'final_assignments': scenario_results[-1]['optimization']['assignments'],
            'kpis_trend': {kpi: [step['kpis'].get(kpi, 0) for step in scenario_results] 
                          for kpi in ['cost_efficiency', 'air_capacity_utilization', 'average_risk_exposure']}
        }
        
        self.simulation_results['scenario_analysis'][scenario_name] = scenario_analysis
        
        return scenario_analysis
    
    def generate_comprehensive_report(self) -> Dict:
        """
        Generate comprehensive digital twin report
        """
        report = {
            'summary': {
                'simulation_steps': self.simulation_step,
                'total_data_points': len(self.simulation_results['sensor_analytics']) * 50,  # Approximate
                'total_alerts': len(self.simulation_results['alerts']),
                'active_scenarios': list(self.simulation_results['scenario_analysis'].keys())
            },
            'performance_metrics': {
                'avg_cost_efficiency': np.mean([kpi.get('cost_efficiency', 0) 
                                              for kpi in self.simulation_results['kpis']]),
                'avg_air_utilization': np.mean([kpi.get('air_capacity_utilization', 0) 
                                              for kpi in self.simulation_results['kpis']]),
                'avg_risk_exposure': np.mean([kpi.get('average_risk_exposure', 0) 
                                            for kpi in self.simulation_results['kpis']]),
                'data_quality_score': self._calculate_data_quality_score()
            },
            'scenario_comparisons': self._compare_scenarios(),
            'recommendations': self._generate_recommendations()
        }
        
        return report
    
    def _calculate_data_quality_score(self) -> float:
        """
        Calculate overall data quality score
        """
        if not self.simulation_results['sensor_analytics']:
            return 0.0
        
        total_readings = sum(analysis['total_readings'] for analysis in self.simulation_results['sensor_analytics'])
        total_anomalies = sum(len(analysis['anomalies']) for analysis in self.simulation_results['sensor_analytics'])
        
        # Data quality score based on anomaly rate (lower is better)
        anomaly_rate = total_anomalies / total_readings if total_readings > 0 else 0
        quality_score = max(0, 1 - anomaly_rate * 10)  # Scale anomaly rate to quality score
        
        return quality_score
    
    def _compare_scenarios(self) -> Dict:
        """
        Compare performance across different scenarios
        """
        comparisons = {}
        
        for scenario_name, analysis in self.simulation_results['scenario_analysis'].items():
            comparisons[scenario_name] = {
                'final_cost': analysis['final_cost'],
                'cost_volatility': np.std(analysis['cost_evolution']),
                'alert_frequency': analysis['alert_frequency'],
                'risk_management': np.mean(analysis['kpis_trend']['average_risk_exposure'])
            }
        
        return comparisons
    
    def _generate_recommendations(self) -> List[str]:
        """
        Generate recommendations based on simulation results
        """
        recommendations = []
        
        # Cost efficiency recommendations
        avg_cost_efficiency = np.mean([kpi.get('cost_efficiency', 0) for kpi in self.simulation_results['kpis']])
        if avg_cost_efficiency > 0.15:  # Cost > 15% of cargo value
            recommendations.append("High cost efficiency detected - consider renegotiating carrier contracts")
        
        # Capacity utilization recommendations
        avg_air_utilization = np.mean([kpi.get('air_capacity_utilization', 0) for kpi in self.simulation_results['kpis']])
        if avg_air_utilization > 0.9:
            recommendations.append("Air capacity consistently high - consider securing additional capacity")
        elif avg_air_utilization < 0.3:
            recommendations.append("Low air capacity utilization - review pricing and demand forecasting")
        
        # Risk management recommendations
        avg_risk_exposure = np.mean([kpi.get('average_risk_exposure', 0) for kpi in self.simulation_results['kpis']])
        if avg_risk_exposure > 0.6:
            recommendations.append("High risk exposure detected - strengthen disruption mitigation strategies")
        
        # Data quality recommendations
        data_quality = self._calculate_data_quality_score()
        if data_quality < 0.8:
            recommendations.append("Data quality concerns detected - investigate sensor maintenance needs")
        
        return recommendations

# Initialize digital twin simulation
digital_twin = DigitalTwinSimulation(shipments, base_params, twin_components)
print("Digital Twin Simulation initialized")
print(f"  Components: IoT sensors, Market feeds, Analytics engine")
print(f"  Simulation ready: Real-time processing and optimization")
```

```python
# Demonstrate predictive analytics and what-if analysis
def demonstrate_predictive_capabilities(digital_twin: DigitalTwinSimulation) -> Dict:
    """
    Demonstrate the predictive analytics and what-if analysis capabilities
    """
    print("\nDemonstrating Predictive Analytics and What-If Analysis...")
    
    # Generate multiple what-if scenarios
    what_if_scenarios = {
        'red_sea_crisis': {
            'description': 'Red Sea crisis causing 40% ocean rate increase and 50% delay probability',
            'market_adjustments': {
                'ocean_rate_index': 1.4,
                'air_rate_index': 1.15,
                'weather_risk': {dest: 0.5 for dest in digital_twin.market_feed.destinations}
            }
        },
        'pandemic_recovery': {
            'description': 'Post-pandemic recovery with high demand and capacity constraints',
            'market_adjustments': {
                'ocean_rate_index': 1.25,
                'air_rate_index': 1.4,
                'port_congestion': {dest: 0.85 for dest in digital_twin.market_feed.destinations}
            }
        },
        'fuel_price_shock': {
            'description': 'Fuel price shock affecting both ocean and air freight costs',
            'market_adjustments': {
                'fuel_price_index': 2.0,
                'ocean_rate_index': 1.35,
                'air_rate_index': 1.6
            }
        },
        'perfect_conditions': {
            'description': 'Ideal market conditions with low rates and no disruptions',
            'market_adjustments': {
                'ocean_rate_index': 0.85,
                'air_rate_index': 0.9,
                'port_congestion': {dest: 0.2 for dest in digital_twin.market_feed.destinations},
                'weather_risk': {dest: 0.05 for dest in digital_twin.market_feed.destinations}
            }
        }
    }
    
    predictive_results = {}
    
    for scenario_name, scenario_config in what_if_scenarios.items():
        print(f"\nAnalyzing {scenario_name.replace('_', ' ').title()} scenario...")
        
        # Apply scenario conditions
        digital_twin.market_feed.base_conditions.update(scenario_config['market_adjustments'])
        
        # Run simulation
        scenario_results = []
        for step in range(12):  # 12 steps for what-if analysis
            step_result = digital_twin.run_simulation_step()
            scenario_results.append(step_result)
        
        # Analyze results
        final_cost = scenario_results[-1]['optimization']['total_cost']
        cost_trend = [step['optimization']['total_cost'] for step in scenario_results]
        
        # Calculate key metrics
        air_assignments = sum(1 for assignment in scenario_results[-1]['optimization']['assignments'].values() 
                           if assignment['mode'] == 'air')
        air_percentage = air_assignments / len(digital_twin.shipments) * 100
        
        avg_risk = np.mean([assignment['disruption_probability'] 
                           for assignment in scenario_results[-1]['optimization']['assignments'].values()])
        
        predictive_results[scenario_name] = {
            'description': scenario_config['description'],
            'final_cost': final_cost,
            'cost_trend': cost_trend,
            'air_usage_percentage': air_percentage,
            'average_risk': avg_risk,
            'adaptation_speed': self._calculate_adaptation_speed(cost_trend),
            'recommendations': self._generate_scenario_recommendations(scenario_results)
        }
        
        print(f"  Final cost: ${final_cost:,.0f}")
        print(f"  Air freight usage: {air_percentage:.1f}%")
        print(f"  Average risk exposure: {avg_risk:.1%}")
        print(f"  Adaptation speed: {predictive_results[scenario_name]['adaptation_speed']:.2f}")
    
    return predictive_results
    
def _calculate_adaptation_speed(self, cost_trend: List[float]) -> float:
    """
    Calculate how quickly the system adapts to new conditions
    """
    if len(cost_trend) < 3:
        return 0.0
    
    # Calculate rate of change stabilization
    changes = [abs(cost_trend[i+1] - cost_trend[i]) for i in range(len(cost_trend)-1)]
    
    # Adaptation speed: lower volatility in later steps indicates faster adaptation
    early_volatility = np.mean(changes[:len(changes)//2]) if len(changes) > 2 else changes[0]
    late_volatility = np.mean(changes[len(changes)//2:]) if len(changes) > 2 else changes[-1]
    
    adaptation_speed = max(0, 1 - (late_volatility / early_volatility)) if early_volatility > 0 else 1.0
    return adaptation_speed
    
def _generate_scenario_recommendations(self, scenario_results: List[Dict]) -> List[str]:
    """
    Generate recommendations based on scenario results
    """
    recommendations = []
    
    # Analyze final state
    final_state = scenario_results[-1]
    final_assignments = final_state['optimization']['assignments']
    
    # Air capacity analysis
    air_assignments = sum(1 for assignment in final_assignments.values() if assignment['mode'] == 'air')
    air_percentage = air_assignments / len(final_assignments) * 100
    
    if air_percentage > 70:
        recommendations.append("High air freight utilization - secure additional air capacity contracts")
    elif air_percentage < 20:
        recommendations.append("Low air freight usage - opportunity for cost optimization")
    
    # Risk analysis
    high_risk_shipments = sum(1 for assignment in final_assignments.values() 
                             if assignment['disruption_probability'] > 0.7)
    
    if high_risk_shipments > len(final_assignments) * 0.5:
        recommendations.append("High risk exposure - implement additional mitigation strategies")
    
    # Cost analysis
    cost_trend = [step['optimization']['total_cost'] for step in scenario_results]
    if len(cost_trend) > 1:
        cost_increase = (cost_trend[-1] - cost_trend[0]) / cost_trend[0] * 100
        if cost_increase > 20:
            recommendations.append("Significant cost increase - review carrier contracts and routing options")
    
    return recommendations

# Run predictive analytics demonstration
predictive_results = demonstrate_predictive_capabilities(digital_twin)

# Create predictive analytics visualization
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Digital Twin Predictive Analytics & What-If Analysis', fontsize=16, fontweight='bold')

# 1. Cost comparison across what-if scenarios
scenario_names = list(predictive_results.keys())
final_costs = [predictive_results[name]['final_cost'] for name in scenario_names]
air_usage = [predictive_results[name]['air_usage_percentage'] for name in scenario_names]
risk_levels = [predictive_results[name]['average_risk'] * 100 for name in scenario_names]

x = np.arange(len(scenario_names))
width = 0.25

bars1 = ax1.bar(x - width, final_costs, width, label='Final Cost', alpha=0.8, color='steelblue')
ax1_twin = ax1.twinx()
bars2 = ax1_twin.bar(x, air_usage, width, label='Air Usage %', alpha=0.8, color='orange')
bars3 = ax1_twin.bar(x + width, risk_levels, width, label='Risk Level %', alpha=0.8, color='red')

ax1.set_xlabel('What-If Scenario')
ax1.set_ylabel('Cost ($)', color='steelblue')
ax1_twin.set_ylabel('Percentage', color='orange')
ax1.set_title('What-If Scenario Comparison')
ax1.set_xticks(x)
ax1.set_xticklabels([s.replace('_', '\n').title() for s in scenario_names], rotation=45)
ax1.tick_params(axis='y', labelcolor='steelblue')
ax1_twin.tick_params(axis='y', labelcolor='orange')
ax1.legend(loc='upper left')
ax1_twin.legend(loc='upper right')
ax1.grid(True, alpha=0.3)

# 2. Adaptation speed analysis
adaptation_speeds = [predictive_results[name]['adaptation_speed'] for name in scenario_names]
colors = ['green' if speed > 0.7 else 'orange' if speed > 0.4 else 'red' for speed in adaptation_speeds]

bars = ax2.bar(scenario_names, adaptation_speeds, color=colors, alpha=0.7)
ax2.set_xlabel('Scenario')
ax2.set_ylabel('Adaptation Speed')
ax2.set_title('System Adaptation Speed by Scenario')
ax2.set_ylim(0, 1)
ax2.tick_params(axis='x', rotation=45)
ax2.grid(True, alpha=0.3)
ax2.axhline(y=0.7, color='green', linestyle='--', label='Fast Adaptation', linewidth=2)
ax2.axhline(y=0.4, color='orange', linestyle='--', label='Medium Adaptation', linewidth=2)
ax2.legend()

# Add speed labels
for bar, speed in zip(bars, adaptation_speeds):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 0.02,
             f'{speed:.2f}', ha='center', va='bottom', fontweight='bold')

# 3. Cost trend analysis for Red Sea crisis
if 'red_sea_crisis' in predictive_results:
    cost_trend = predictive_results['red_sea_crisis']['cost_trend']
    steps = range(len(cost_trend))
    
    ax3.plot(steps, cost_trend, 'r-o', linewidth=2, markersize=6, label='Cost Evolution')
    
    # Add trend line
    z = np.polyfit(steps, cost_trend, 1)
    p = np.poly1d(z)
    ax3.plot(steps, p(steps), 'r--', alpha=0.5, label='Trend Line')
    
    ax3.set_xlabel('Simulation Step')
    ax3.set_ylabel('Total Cost ($)')
    ax3.set_title('Red Sea Crisis: Cost Evolution & Adaptation')
    ax3.legend()
    ax3.grid(True, alpha=0.3)

# 4. Recommendations summary
all_recommendations = []
for scenario_name, results in predictive_results.items():
    for rec in results['recommendations']:
        all_recommendations.append(f"{scenario_name}: {rec}")

# Create word cloud-like visualization of recommendations
from collections import Counter
recommendation_words = []
for rec in all_recommendations[:10]:  # First 10 recommendations
    words = rec.lower().split()
    recommendation_words.extend([word for word in words if len(word) > 3])

word_counts = Counter(recommendation_words)
top_words = word_counts.most_common(8)

if top_words:
    words, counts = zip(*top_words)
    bars = ax4.bar(words, counts, color='lightgreen', alpha=0.7)
    ax4.set_xlabel('Key Recommendation Terms')
    ax4.set_ylabel('Frequency')
    ax4.set_title('Predictive Analytics: Key Recommendation Themes')
    ax4.tick_params(axis='x', rotation=45)
    ax4.grid(True, alpha=0.3)
else:
    ax4.text(0.5, 0.5, 'No Recommendations\nAvailable', ha='center', va='center', 
             transform=ax4.transAxes, fontsize=14)
    ax4.set_title('Predictive Analytics: Recommendations')

plt.tight_layout()
plt.show()

print("\nPredictive Analytics Summary:")
print("=" * 80)
print(f"Analyzed {len(predictive_results)} what-if scenarios")
print(f"Average adaptation speed: {np.mean(adaptation_speeds):.2f}")
print(f"Cost range: ${min(final_costs):,.0f} - ${max(final_costs):,.0f}")
print(f"Air usage range: {min(air_usage):.1f}% - {max(air_usage):.1f}%")
print(f"Total recommendations generated: {len(all_recommendations)}")
```

### Why this Tier exists vs earlier Tiers
The Digital Twin paradigm represents a fundamental shift from static optimization to dynamic, integrated decision-making:
- **Real-time integration**: Continuous data feeds from physical operations
- **Predictive analytics**: Forecast disruptions before they occur
- **System-of-systems**: Capture complex interdependencies across logistics network
- **What-if analysis**: Test strategies without real-world consequences
- **Continuous learning**: System improves with more data and experience
- **Holistic optimization**: Consider multiple objectives simultaneously

### Pros / Cons vs Tiers 1-4
**Pros:**
- Real-time adaptation to changing market conditions
- Predictive capabilities for proactive decision-making
- Comprehensive system-wide optimization
- Risk mitigation and resilience building
- Continuous improvement through data analytics
- Scenario testing without operational risk
- Multi-objective optimization (cost, service, sustainability)

**Cons:**
- High implementation complexity and cost
- Requires extensive IoT infrastructure and data integration
- Significant computational resources for real-time processing
- Data quality and security concerns
- Complex maintenance and calibration requirements
- Dependence on sensor reliability and data accuracy
- Organizational change management challenges

### When to use this Tier
- **Large-scale operations** with complex global supply chains
- **High-value cargo** where disruptions have significant financial impact
- **Regulatory environments** requiring comprehensive tracking and reporting
- **Strategic planning** requiring long-term scenario analysis
- **Risk-sensitive industries** requiring proactive disruption management
- **Digital transformation initiatives** with IoT and analytics investments
- **Customer service excellence** programs requiring real-time visibility
- **Sustainability reporting** requiring comprehensive environmental impact tracking
