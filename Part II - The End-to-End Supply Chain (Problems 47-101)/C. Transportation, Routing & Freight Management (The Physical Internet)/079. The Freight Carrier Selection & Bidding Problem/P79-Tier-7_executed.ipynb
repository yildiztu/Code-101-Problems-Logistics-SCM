# 79. The Freight Carrier Selection & Bidding Problem
## Tier 7: The Human-AI Symbiotic Partnership (The Centaur Model)

### Key assumptions
- Human expertise and AI analytics complement each other effectively
- Explainable AI provides transparent reasoning for recommendations
- Human strategic insights enhance AI's quantitative analysis
- Trust and confidence metrics guide human-AI collaboration

### Approach (step-by-step)
1. **Integrate AI analysis** with human expert knowledge and experience
2. **Create explainable recommendations** with transparent reasoning
3. **Implement trust scoring** based on AI confidence and human expertise
4. **Facilitate interactive decision refinement** through human feedback
5. **Calculate combined scores** balancing AI and human inputs
6. **Measure partnership benefits** vs AI-only or human-only approaches

### What to look for in the results
- Improved decision quality through human-AI collaboration
- Trust and confidence metrics in AI recommendations
- Explainable reasoning behind carrier selection decisions
- Measurable benefits of the centaur model approach

### Concrete example (from the source)
Human-AI partnership with 3 carriers:
- Expected Winner: Carrier 1 - Final Score: 0.847
- Expected AI Component: 0.732 (strong performance metrics)
- Expected Human Component: 0.875 (excellent relationship, strategic value)
- Expected Risk Adjustment: 1.1 (low risk assessment)
- Expected Confidence Level: 0.895

```python
from typing import Tuple, List, Dict, Set
from itertools import product

# Import required libraries for human-AI symbiotic partnership
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass
from typing import List, Dict, Tuple, Optional
import random
import warnings
warnings.filterwarnings('ignore')

# Set up visualization style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Set random seed for reproducible results
np.random.seed(42)
random.seed(42)
```

```python
try:
    # Define human expert and AI system classes
    
    @dataclass
    class HumanExpert:
        """
        Human logistics expert with domain knowledge and relationship insights.
        """
        id: int
        name: str
        experience_years: int
        expertise_areas: List[str]
        relationship_knowledge: Dict[str, float]  # Carrier relationship scores
        strategic_insights: Dict[str, str]  # Strategic considerations
        risk_tolerance: float  # 0-1, conservative to aggressive
        confidence_level: float  # 0-1, self-assessed confidence
        
        def evaluate_carrier_human(self, carrier_name: str, carrier_data: Dict) -> Dict:
            """
            Human expert evaluation based on experience and relationships.
            """
            
            # Relationship score (based on historical interactions)
            relationship_score = self.relationship_knowledge.get(carrier_name, 0.5)
            
            # Strategic value assessment
            strategic_factors = {
                'market_position': 0.2,  # Carrier's market position
                'innovation_capability': 0.15,  # Technology and process innovation
                'flexibility': 0.15,  # Adaptability to changes
                'long_term_partnership': 0.2,  # Partnership potential
                'reliability_under_pressure': 0.15,  # Performance under stress
                'cost_transparency': 0.15  # Billing and cost clarity
            }
            
            # Simulate human evaluation based on experience
            strategic_score = 0
            for factor, weight in strategic_factors.items():
                # Human evaluation with some randomness and bias based on experience
                base_score = np.random.normal(0.7, 0.1)
                experience_boost = min(0.1, self.experience_years / 100)
                factor_score = np.clip(base_score + experience_boost, 0.3, 1.0)
                strategic_score += factor_score * weight
            
            # Risk assessment based on human judgment
            risk_factors = {
                'financial_stability': np.random.normal(0.8, 0.1),
                'operational_consistency': np.random.normal(0.75, 0.08),
                'market_volatility_exposure': np.random.normal(0.7, 0.12)
            }
            
            risk_assessment = np.mean(list(risk_factors.values()))
            
            # Adjust for expert's risk tolerance
            if self.risk_tolerance < 0.3:  # Risk-averse
                risk_adjustment = 0.9
            elif self.risk_tolerance > 0.7:  # Risk-seeking
                risk_adjustment = 1.1
            else:
                risk_adjustment = 1.0
            
            # Final human score
            human_score = (relationship_score * 0.4 + 
                           strategic_score * 0.4 + 
                           risk_assessment * risk_adjustment * 0.2)
            
            return {
                'human_score': human_score,
                'relationship_score': relationship_score,
                'strategic_score': strategic_score,
                'risk_assessment': risk_assessment,
                'confidence': self.confidence_level,
                'reasoning': self._generate_reasoning(carrier_name, strategic_factors, risk_factors)
            }
        
        def _generate_reasoning(self, carrier_name: str, strategic_factors: Dict, risk_factors: Dict) -> str:
            """
            Generate human-like reasoning for carrier evaluation.
            """
            
            reasoning_points = []
            
            # Relationship-based reasoning
            rel_score = self.relationship_knowledge.get(carrier_name, 0.5)
            if rel_score > 0.8:
                reasoning_points.append(f"Excellent long-term relationship with {carrier_name}")
            elif rel_score > 0.6:
                reasoning_points.append(f"Good working relationship with {carrier_name}")
            elif rel_score < 0.4:
                reasoning_points.append(f"Limited experience with {carrier_name}")
            
            # Strategic reasoning
            if self.experience_years > 15:
                reasoning_points.append(f"Based on {self.experience_years} years of industry experience")
            
            # Risk reasoning
            if self.risk_tolerance < 0.3:
                reasoning_points.append("Prefer conservative, reliable partners")
            elif self.risk_tolerance > 0.7:
                reasoning_points.append("Open to innovative, higher-risk options")
            
            return "; ".join(reasoning_points)
    
    @dataclass
    class AIAnalyst:
        """
        AI system with explainable analytics and confidence scoring.
        """
        id: int
        name: str
        model_version: str
        data_sources: List[str]
        analytical_capabilities: List[str]
        confidence_threshold: float
        
        def __post_init__(self):
            self.historical_data = self._generate_historical_data()
            self.model_accuracy = 0.87  # Overall model accuracy
        
        def _generate_historical_data(self) -> pd.DataFrame:
            """
            Generate synthetic historical performance data.
            """
            
            carriers = ['Carrier 1', 'Carrier 2', 'Carrier 3']
            months = 24
            
            data = []
            for carrier in carriers:
                for month in range(months):
                    data.append({
                        'carrier': carrier,
                        'month': month,
                        'on_time_performance': np.random.normal(0.85, 0.08),
                        'cost_efficiency': np.random.normal(0.78, 0.06),
                        'service_quality': np.random.normal(0.82, 0.07),
                        'compliance_rate': np.random.normal(0.94, 0.04),
                        'flexibility_score': np.random.normal(0.76, 0.09)
                    })
            
            return pd.DataFrame(data)
        
        def analyze_carrier_ai(self, carrier_name: str, carrier_data: Dict) -> Dict:
            """
            AI analysis with explainable reasoning and confidence scoring.
            """
            
            # Get historical performance for this carrier
            carrier_hist = self.historical_data[self.historical_data['carrier'] == carrier_name]
            
            # Calculate performance metrics
            performance_metrics = {
                'on_time_performance': carrier_hist['on_time_performance'].mean(),
                'cost_efficiency': carrier_hist['cost_efficiency'].mean(),
                'service_quality': carrier_hist['service_quality'].mean(),
                'compliance_rate': carrier_hist['compliance_rate'].mean(),
                'flexibility_score': carrier_hist['flexibility_score'].mean()
            }
            
            # Calculate trend analysis (recent 6 months vs overall)
            recent_performance = carrier_hist.tail(6)
            
            trend_metrics = {}
            for metric in performance_metrics:
                recent_avg = recent_performance[metric].mean()
                overall_avg = performance_metrics[metric]
                trend = (recent_avg - overall_avg) / overall_avg
                trend_metrics[f'{metric}_trend'] = trend
            
            # Predictive analytics
            predictive_scores = self._predictive_analysis(carrier_name, performance_metrics, trend_metrics)
            
            # Calculate AI confidence
            data_quality = len(carrier_hist) / 24  # Data completeness
            model_confidence = self.model_accuracy * data_quality
            
            # Calculate AI score
            ai_score = (
                np.mean(list(performance_metrics.values())) * 0.6 +
                np.mean([max(0, v) for v in trend_metrics.values()]) * 0.2 +
                predictive_scores['overall_prediction'] * 0.2
            )
            
            return {
                'ai_score': ai_score,
                'performance_metrics': performance_metrics,
                'trend_metrics': trend_metrics,
                'predictive_scores': predictive_scores,
                'confidence': model_confidence,
                'reasoning': self._generate_ai_reasoning(carrier_name, performance_metrics, trend_metrics)
            }
        
        def _predictive_analysis(self, carrier_name: str, performance_metrics: Dict, trend_metrics: Dict) -> Dict:
            """
            Generate predictive analytics for future performance.
            """
            
            # Simulate predictive model outputs
            base_prediction = np.mean(list(performance_metrics.values()))
            
            # Adjust based on trends
            trend_adjustment = np.mean([max(-0.2, min(0.2, v)) for v in trend_metrics.values()])
            
            predictions = {
                'overall_prediction': np.clip(base_prediction + trend_adjustment, 0.3, 1.0),
                'reliability_prediction': performance_metrics['on_time_performance'] + trend_metrics.get('on_time_performance_trend', 0),
                'cost_prediction': performance_metrics['cost_efficiency'] + trend_metrics.get('cost_efficiency_trend', 0),
                'risk_assessment': 1 - performance_metrics['compliance_rate']
            }
            
            return predictions
        
        def _generate_ai_reasoning(self, carrier_name: str, performance_metrics: Dict, trend_metrics: Dict) -> str:
            """
            Generate explainable AI reasoning.
            """
            
            reasoning_points = []
            
            # Performance-based reasoning
            best_metric = max(performance_metrics, key=performance_metrics.get)
            worst_metric = min(performance_metrics, key=performance_metrics.get)
            
            reasoning_points.append(f"Strongest in {best_metric.replace('_', ' ')}: {performance_metrics[best_metric]:.3f}")
            reasoning_points.append(f"Needs improvement in {worst_metric.replace('_', ' ')}: {performance_metrics[worst_metric]:.3f}")
            
            # Trend-based reasoning
            improving_metrics = [k.replace('_trend', '') for k, v in trend_metrics.items() if v > 0.05]
            if improving_metrics:
                reasoning_points.append(f"Improving trends in: {', '.join(improving_metrics)}")
            
            # Data-based reasoning
            reasoning_points.append(f"Analysis based on {len(self.historical_data[self.historical_data['carrier'] == carrier_name])} months of data")
            
            return "; ".join(reasoning_points)
    
    @dataclass
    class CentaurPartnership:
        """
        Human-AI symbiotic partnership system.
        """
        human_expert: HumanExpert
        ai_analyst: AIAnalyst
        collaboration_model: str  # 'sequential', 'interactive', 'integrated'
        trust_threshold: float
        
        def evaluate_carrier_partnership(self, carrier_name: str, carrier_data: Dict) -> Dict:
            """
            Combined human-AI evaluation with symbiotic decision making.
            """
            
            # Get individual evaluations
            human_eval = self.human_expert.evaluate_carrier_human(carrier_name, carrier_data)
            ai_eval = self.ai_analyst.analyze_carrier_ai(carrier_name, carrier_data)
            
            # Calculate trust score
            trust_score = self._calculate_trust_score(human_eval, ai_eval)
            
            # Adaptive weighting based on trust and confidence
            if trust_score > self.trust_threshold:
                # High trust: more balanced weight
                human_weight = 0.5
                ai_weight = 0.5
            else:
                # Low trust: favor human judgment
                human_weight = 0.7
                ai_weight = 0.3
            
            # Adjust weights based on confidence levels
            human_confidence_factor = human_eval['confidence']
            ai_confidence_factor = ai_eval['confidence']
            
            total_confidence = human_confidence_factor + ai_confidence_factor
            human_weight *= (human_confidence_factor / total_confidence)
            ai_weight *= (ai_confidence_factor / total_confidence)
            
            # Calculate combined score
            combined_score = (human_eval['human_score'] * human_weight + 
                              ai_eval['ai_score'] * ai_weight)
            
            # Risk adjustment
            human_risk = human_eval['risk_assessment']
            ai_risk = ai_eval['predictive_scores']['risk_assessment']
            combined_risk = (human_risk * human_weight + ai_risk * ai_weight)
            
            risk_adjustment = 1.0 - (combined_risk * 0.1)  # Reduce score based on risk
            final_score = combined_score * risk_adjustment
            
            # Calculate overall confidence
            overall_confidence = (human_confidence_factor * human_weight + 
                                 ai_confidence_factor * ai_weight) * trust_score
            
            return {
                'final_score': final_score,
                'human_score': human_eval['human_score'],
                'ai_score': ai_eval['ai_score'],
                'human_weight': human_weight,
                'ai_weight': ai_weight,
                'trust_score': trust_score,
                'risk_assessment': combined_risk,
                'risk_adjustment': risk_adjustment,
                'overall_confidence': overall_confidence,
                'human_reasoning': human_eval['reasoning'],
                'ai_reasoning': ai_eval['reasoning'],
                'partnership_recommendation': self._generate_partnership_recommendation(final_score, overall_confidence)
            }
        
        def _calculate_trust_score(self, human_eval: Dict, ai_eval: Dict) -> float:
            """
            Calculate trust score between human and AI evaluations.
            """
            
            # Score alignment
            score_diff = abs(human_eval['human_score'] - ai_eval['ai_score'])
            alignment_score = 1.0 - min(score_diff, 1.0)
            
            # Confidence alignment
            confidence_diff = abs(human_eval['confidence'] - ai_eval['confidence'])
            confidence_alignment = 1.0 - min(confidence_diff, 1.0)
            
            # Risk assessment alignment
            human_risk = human_eval['risk_assessment']
            ai_risk = ai_eval['predictive_scores']['risk_assessment']
            risk_diff = abs(human_risk - ai_risk)
            risk_alignment = 1.0 - min(risk_diff, 1.0)
            
            # Overall trust score
            trust_score = (alignment_score * 0.4 + confidence_alignment * 0.3 + risk_alignment * 0.3)
            
            return trust_score
        
        def _generate_partnership_recommendation(self, final_score: float, confidence: float) -> str:
            """
            Generate partnership recommendation based on combined evaluation.
            """
            
            if final_score > 0.8 and confidence > 0.8:
                return "STRONG RECOMMEND - High confidence, excellent fit"
            elif final_score > 0.7 and confidence > 0.6:
                return "RECOMMEND - Good fit with reasonable confidence"
            elif final_score > 0.6 and confidence > 0.5:
                return "CONSIDER - Acceptable option, monitor closely"
            elif final_score > 0.5:
                return "CAUTION - Below threshold, specific circumstances only"
            else:
                return "AVOID - High risk, poor fit"
    
    # Initialize the human-AI partnership system
    def initialize_centaur_system():
        """
        Create and initialize the centaur partnership system.
        """
        
        # Create human expert
        human_expert = HumanExpert(
            id=1,
            name="Sarah Chen",
            experience_years=18,
            expertise_areas=["Carrier Relations", "Risk Management", "Strategic Sourcing"],
            relationship_knowledge={
                "Carrier 1": 0.85,  # Excellent relationship
                "Carrier 2": 0.65,  # Good relationship
                "Carrier 3": 0.45   # Limited experience
            },
            strategic_insights={
                "Carrier 1": "Innovative, flexible, strong growth potential",
                "Carrier 2": "Reliable, conservative, stable performance",
                "Carrier 3": "Cost-effective, expanding service area"
            },
            risk_tolerance=0.4,  # Slightly risk-averse
            confidence_level=0.82
        )
        
        # Create AI analyst
        ai_analyst = AIAnalyst(
            id=1,
            name="LogisticsAI-2.0",
            model_version="v2.3.1",
            data_sources=["Historical Performance", "Market Data", "Financial Reports", "Customer Feedback"],
            analytical_capabilities=["Predictive Analytics", "Risk Assessment", "Performance Optimization", "Trend Analysis"],
            confidence_threshold=0.75
        )
        
        # Create centaur partnership
        partnership = CentaurPartnership(
            human_expert=human_expert,
            ai_analyst=ai_analyst,
            collaboration_model="integrated",
            trust_threshold=0.7
        )
        
        return partnership
    
    # Initialize centaur system
    centaur_system = initialize_centaur_system()
    
    print("Human-AI Symbiotic Partnership System Initialized:")
    print(f"Human Expert: {centaur_system.human_expert.name} ({centaur_system.human_expert.experience_years} years experience)")
    print(f"AI Analyst: {centaur_system.ai_analyst.name} ({centaur_system.ai_analyst.model_version})")
    print(f"Collaboration Model: {centaur_system.collaboration_model}")
    print(f"Trust Threshold: {centaur_system.trust_threshold:.1f}")
    
    print(f"\nHuman Expert Expertise: {', '.join(centaur_system.human_expert.expertise_areas)}")
    print(f"AI Capabilities: {', '.join(centaur_system.ai_analyst.analytical_capabilities)}")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    # Apply centaur partnership to carrier selection problem
    
    def centaur_carrier_selection():
        """
        Apply human-AI symbiotic partnership to carrier selection.
        """
        
        # Define current bidding scenario
        carriers_data = {
            'Carrier 1': {
                'bid_price': 500,
                'capacity': 150,
                'reliability': 95,
                'service_quality': 90,
                'coverage': 'National',
                'technology_level': 'Advanced'
            },
            'Carrier 2': {
                'bid_price': 480,
                'capacity': 120,
                'reliability': 88,
                'service_quality': 92,
                'coverage': 'Regional',
                'technology_level': 'Moderate'
            },
            'Carrier 3': {
                'bid_price': 520,
                'capacity': 200,
                'reliability': 92,
                'service_quality': 87,
                'coverage': 'International',
                'technology_level': 'Advanced'
            }
        }
        
        # Evaluate each carrier using centaur partnership
        centaur_evaluations = {}
        
        print("Human-AI Partnership Carrier Evaluation:")
        print("=" * 60)
        
        for carrier_name, carrier_data in carriers_data.items():
            print(f"\n🔍 Evaluating {carrier_name}:")
            
            evaluation = centaur_system.evaluate_carrier_partnership(carrier_name, carrier_data)
            centaur_evaluations[carrier_name] = evaluation
            
            print(f"   Final Score: {evaluation['final_score']:.3f}")
            print(f"   Human Score: {evaluation['human_score']:.3f} (Weight: {evaluation['human_weight']:.2f})")
            print(f"   AI Score: {evaluation['ai_score']:.3f} (Weight: {evaluation['ai_weight']:.2f})")
            print(f"   Trust Score: {evaluation['trust_score']:.3f}")
            print(f"   Risk Assessment: {evaluation['risk_assessment']:.3f}")
            print(f"   Overall Confidence: {evaluation['overall_confidence']:.3f}")
            print(f"   Recommendation: {evaluation['partnership_recommendation']}")
            
            print(f"   \n   🧠 Human Reasoning: {evaluation['human_reasoning']}")
            print(f"   🤖 AI Reasoning: {evaluation['ai_reasoning']}")
        
        return centaur_evaluations, carriers_data
    
    # Run centaur carrier selection
    centaur_evaluations, carriers_data = centaur_carrier_selection()
    
    # Make final selection based on centaur evaluations
    best_carrier = max(centaur_evaluations, key=lambda x: centaur_evaluations[x]['final_score'])
    best_evaluation = centaur_evaluations[best_carrier]
    
    print(f"\n" + "=" * 60)
    print(f"CENTAUR PARTNERSHIP FINAL DECISION")
    print(f"=" * 60)
    print(f"\n🏆 WINNER: {best_carrier}")
    print(f"   Final Score: {best_evaluation['final_score']:.3f}")
    print(f"   Overall Confidence: {best_evaluation['overall_confidence']:.3f}")
    print(f"   Recommendation: {best_evaluation['partnership_recommendation']}")
    
    # Calculate total cost
    demand = [100, 80]  # Lane 1 and Lane 2 demand
    total_cost = carriers_data[best_carrier]['bid_price'] * sum(demand)
    
    print(f"   Total Cost: ${total_cost:,.2f}")
    print(f"   Risk Adjustment Factor: {best_evaluation['risk_adjustment']:.3f}")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    # Comprehensive visualization of centaur partnership results
    
    def visualize_centaur_results(evaluations: Dict, carriers_data: Dict):
        """
        Create comprehensive visualizations of the centaur partnership results.
        """
        
        fig, axes = plt.subplots(2, 2, figsize=(15, 12))
        fig.suptitle('Human-AI Symbiotic Partnership Analysis', fontsize=16, fontweight='bold')
        
        # Prepare data for visualization
        carrier_names = list(evaluations.keys())
        final_scores = [evaluations[name]['final_score'] for name in carrier_names]
        human_scores = [evaluations[name]['human_score'] for name in carrier_names]
        ai_scores = [evaluations[name]['ai_score'] for name in carrier_names]
        trust_scores = [evaluations[name]['trust_score'] for name in carrier_names]
        confidence_scores = [evaluations[name]['overall_confidence'] for name in carrier_names]
        
        # 1. Score comparison chart
        ax1 = axes[0, 0]
        x = np.arange(len(carrier_names))
        width = 0.25
        
        bars1 = ax1.bar(x - width, human_scores, width, label='Human Score', color='lightblue', alpha=0.8)
        bars2 = ax1.bar(x, ai_scores, width, label='AI Score', color='lightgreen', alpha=0.8)
        bars3 = ax1.bar(x + width, final_scores, width, label='Final Score', color='gold', alpha=0.8)
        
        ax1.set_xlabel('Carriers')
        ax1.set_ylabel('Score')
        ax1.set_title('Human vs AI vs Final Score Comparison')
        ax1.set_xticks(x)
        ax1.set_xticklabels(carrier_names)
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        ax1.set_ylim(0, 1)
        
        # 2. Trust and confidence analysis
        ax2 = axes[0, 1]
        
        # Create scatter plot of trust vs confidence
        scatter = ax2.scatter(trust_scores, confidence_scores, 
                           s=200, c=final_scores, cmap='viridis', 
                           alpha=0.7, edgecolors='black', linewidth=2)
        
        # Add carrier labels
        for i, name in enumerate(carrier_names):
            ax2.annotate(name, (trust_scores[i], confidence_scores[i]), 
                        xytext=(5, 5), textcoords='offset points', fontweight='bold')
        
        ax2.set_xlabel('Trust Score')
        ax2.set_ylabel('Overall Confidence')
        ax2.set_title('Trust-Confidence Relationship')
        ax2.grid(True, alpha=0.3)
        ax2.set_xlim(0, 1)
        ax2.set_ylim(0, 1)
        
        # Add colorbar for final scores
        cbar = plt.colorbar(scatter, ax=ax2)
        cbar.set_label('Final Score')
        
        # 3. Weight distribution analysis
        ax3 = axes[1, 0]
        
        human_weights = [evaluations[name]['human_weight'] for name in carrier_names]
        ai_weights = [evaluations[name]['ai_weight'] for name in carrier_names]
        
        bars = ax3.bar(carrier_names, human_weights, label='Human Weight', 
                       color='lightblue', alpha=0.8)
        ax3.bar(carrier_names, ai_weights, bottom=human_weights, 
               label='AI Weight', color='lightgreen', alpha=0.8)
        
        ax3.set_ylabel('Decision Weight')
        ax3.set_title('Human vs AI Decision Weights')
        ax3.legend()
        ax3.grid(True, alpha=0.3)
        ax3.set_ylim(0, 1)
        
        # Add weight percentages on bars
        for i, name in enumerate(carrier_names):
            human_weight = human_weights[i]
            ai_weight = ai_weights[i]
            ax3.text(i, human_weight/2, f'{human_weight:.1%}', ha='center', va='center', fontweight='bold')
            ax3.text(i, human_weight + ai_weight/2, f'{ai_weight:.1%}', ha='center', va='center', fontweight='bold')
        
        # 4. Risk assessment and cost analysis
        ax4 = axes[1, 1]
        
        risk_assessments = [evaluations[name]['risk_assessment'] for name in carrier_names]
        bid_prices = [carriers_data[name]['bid_price'] for name in carrier_names]
        
        # Create bubble chart: cost (x), risk (y), final score (size)
        bubble_sizes = [score * 1000 for score in final_scores]  # Scale for visibility
        
        scatter = ax4.scatter(bid_prices, risk_assessments, s=bubble_sizes, 
                           c=final_scores, cmap='plasma', alpha=0.7, 
                           edgecolors='black', linewidth=2)
        
        # Add carrier labels
        for i, name in enumerate(carrier_names):
            ax4.annotate(name, (bid_prices[i], risk_assessments[i]), 
                        xytext=(5, 5), textcoords='offset points', fontweight='bold')
        
        ax4.set_xlabel('Bid Price ($)')
        ax4.set_ylabel('Risk Assessment')
        ax4.set_title('Cost-Risk-Value Analysis')
        ax4.grid(True, alpha=0.3)
        
        # Add colorbar for final scores
        cbar = plt.colorbar(scatter, ax=ax4)
        cbar.set_label('Final Score')
        
        plt.tight_layout()
        plt.show()
    
    # Visualize centaur results
    visualize_centaur_results(centaur_evaluations, carriers_data)
except Exception as e:
    print(f'Error in cell: {e}')
```
