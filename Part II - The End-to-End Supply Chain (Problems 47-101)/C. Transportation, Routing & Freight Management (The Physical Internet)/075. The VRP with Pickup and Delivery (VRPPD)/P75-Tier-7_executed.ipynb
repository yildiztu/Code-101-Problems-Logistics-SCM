```python
try:
    @dataclass
    class VRPPDInstance:
        """Data structure for VRPPD problem instance"""
        num_pairs: int
        num_vehicles: int
        vehicle_capacity: int
        distances: np.ndarray
        demands: List[int]
        
        def __post_init__(self):
            self.num_vertices = 2 * self.num_pairs + 2
            self.depot_start = 0
            self.depot_end = 2 * self.num_pairs + 1
            self.pickups = list(range(1, self.num_pairs + 1))
            self.deliveries = list(range(self.num_pairs + 1, 2 * self.num_pairs + 1))
            self.all_vertices = list(range(self.num_vertices))
            self.vehicles = list(range(self.num_vehicles))
    
    @dataclass
    class DecisionExplanation:
        """Explanation for AI decision making"""
        decision_id: str
        decision_type: str  # 'route_assignment', 'vehicle_selection', 'priority_adjustment'
        reasoning: List[str]  # Step-by-step reasoning process
        key_factors: Dict[str, float]  # Factors influencing decision with weights
        alternatives: List[Dict[str, Any]]  # Alternative options considered
        confidence: float  # Confidence in the decision (0-1)
        timestamp: float
    
    @dataclass
    class HumanFeedback:
        """Human feedback on AI decisions"""
        feedback_id: str
        decision_id: str
        human_id: str
        rating: float  # 1-5 rating of decision quality
        comments: str  # Human comments and suggestions
        corrections: Dict[str, Any]  # Human-suggested corrections
        trust_impact: float  # Impact on trust score (-1 to +1)
        timestamp: float
    
    @dataclass
    class TrustMetrics:
        """Trust metrics between human and AI"""
        human_id: str
        trust_score: float  # 0-1 overall trust level
        competence_trust: float  # Trust in AI competence
        integrity_trust: float  # Trust in AI reliability
        benevolence_trust: float  # Trust in AI intentions
        interaction_count: int  # Number of interactions
        satisfaction_history: List[float]  # Satisfaction scores over time
    
    # Create instance for human-AI partnership demonstration
    def create_partnership_instance():
        """Create a 10-pair instance suitable for human-AI partnership"""
        num_pairs = 10
        num_vehicles = 6
        vehicle_capacity = 10
        
        # Generate realistic demands
        np.random.seed(42)
        pickup_demands = np.random.randint(1, 5, num_pairs).tolist()
        delivery_demands = [-d for d in pickup_demands]
        demands = pickup_demands + delivery_demands
        
        # Generate distance matrix
        num_vertices = 2 * num_pairs + 2
        distances = np.zeros((num_vertices, num_vertices))
        
        # Create realistic distance structure
        for i in range(num_vertices):
            for j in range(num_vertices):
                if i != j:
                    # Base distance with geographic structure
                    base_dist = 8 + abs(i - j) * 1.5 + np.random.uniform(-1, 1)
                    distances[i][j] = max(1, base_dist)
        
        # Make symmetric
        distances = (distances + distances.T) / 2
        np.fill_diagonal(distances, 0)
        
        return VRPPDInstance(num_pairs, num_vehicles, vehicle_capacity, distances, demands)
    
    # Create the partnership instance
    partnership_instance = create_partnership_instance()
    print(f"VRPPD Human-AI Partnership Instance created:")
    print(f"- Pickup-delivery pairs: {partnership_instance.num_pairs}")
    print(f"- Vehicles: {partnership_instance.num_vehicles}")
    print(f"- Vehicle capacity: {partnership_instance.vehicle_capacity}")
    print(f"- Total vertices: {partnership_instance.num_vertices}")
    print(f"- Total demand: {sum(abs(d) for d in partnership_instance.demands)}")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
class ExplainableAI:
    """Explainable AI system for VRPPD decision making"""
    
    def __init__(self, instance: VRPPDInstance):
        self.instance = instance
        self.decision_history = []
        self.explanation_templates = {
            'route_assignment': [
                "Assigning request {request_id} to vehicle {vehicle_id}",
                "Current vehicle load: {current_load}/{capacity}",
                "Distance from current position: {distance:.1f} km",
                "Time window compatibility: {time_window_status}",
                "Alternative vehicles considered: {alternatives_count}"
            ],
            'priority_adjustment': [
                "Adjusting priority for request {request_id}",
                "Original priority: {old_priority}",
                "New priority: {new_priority}",
                "Reason: {adjustment_reason}",
                "Impact on overall schedule: {schedule_impact}"
            ],
            'route_optimization': [
                "Optimizing route for vehicle {vehicle_id}",
                "Current route length: {current_length:.1f} km",
                "Optimization potential: {potential_savings:.1f} km",
                "Constraints considered: {constraints}",
                "Optimization method: {method}"
            ]
        }
        self.confidence_threshold = 0.7
    
    def calculate_route_distance(self, route: List[int]) -> float:
        """Calculate total distance of a route"""
        if len(route) < 2:
            return 0.0
        total_distance = 0.0
        for i in range(len(route) - 1):
            total_distance += self.instance.distances[route[i]][route[i + 1]]
        return total_distance
    
    def make_decision_with_explanation(self, decision_type: str, context: Dict[str, Any]) -> Tuple[Dict[str, Any], DecisionExplanation]:
        """Make AI decision with detailed explanation"""
        
        decision_id = f"decision_{len(self.decision_history)}_{int(time.time())}"
        
        if decision_type == 'route_assignment':
            decision, explanation = self._assign_route_with_explanation(context, decision_id)
        elif decision_type == 'priority_adjustment':
            decision, explanation = self._adjust_priority_with_explanation(context, decision_id)
        elif decision_type == 'route_optimization':
            decision, explanation = self._optimize_route_with_explanation(context, decision_id)
        else:
            decision = {'status': 'unknown_decision_type'}
            explanation = DecisionExplanation(
                decision_id=decision_id,
                decision_type=decision_type,
                reasoning=["Unknown decision type"],
                key_factors={},
                alternatives=[],
                confidence=0.0,
                timestamp=time.time()
            )
        
        self.decision_history.append({
            'decision_id': decision_id,
            'decision': decision,
            'explanation': explanation,
            'context': context
        })
        
        return decision, explanation
    
    def _assign_route_with_explanation(self, context: Dict[str, Any], decision_id: str) -> Tuple[Dict[str, Any], DecisionExplanation]:
        """Assign route with detailed explanation"""
        
        request_id = context['request_id']
        pickup_location = context['pickup_location']
        delivery_location = context['delivery_location']
        demand = context['demand']
        current_vehicle_states = context['vehicle_states']
        
        # Find best vehicle assignment
        best_vehicle = None
        best_score = float('inf')
        alternatives = []
        
        for vehicle_id, state in current_vehicle_states.items():
            if state['load'] + demand <= self.instance.vehicle_capacity:
                # Calculate assignment score
                distance_to_pickup = self.instance.distances[state['current_position']][pickup_location]
                pickup_to_delivery = self.instance.distances[pickup_location][delivery_location]
                total_distance = distance_to_pickup + pickup_to_delivery
                
                score = total_distance * (1 + state['load'] / self.instance.vehicle_capacity * 0.5)
                
                alternatives.append({
                    'vehicle_id': vehicle_id,
                    'score': score,
                    'distance': total_distance,
                    'load_after': state['load'] + demand
                })
                
                if score < best_score:
                    best_score = score
                    best_vehicle = vehicle_id
        
        # Sort alternatives by score
        alternatives.sort(key=lambda x: x['score'])
        
        if best_vehicle is not None:
            decision = {
                'action': 'assign_request',
                'request_id': request_id,
                'vehicle_id': best_vehicle,
                'confidence': max(0.5, 1.0 - (len(alternatives) - 1) * 0.1)
            }
            
            # Generate explanation
            best_state = current_vehicle_states[best_vehicle]
            reasoning = []
            
            for template in self.explanation_templates['route_assignment']:
                if '{request_id}' in template:
                    reasoning.append(template.format(request_id=request_id))
                elif '{vehicle_id}' in template:
                    reasoning.append(template.format(vehicle_id=best_vehicle))
                elif '{current_load}' in template:
                    reasoning.append(template.format(current_load=best_state['load'], capacity=self.instance.vehicle_capacity))
                elif '{distance}' in template:
                    distance_to_pickup = self.instance.distances[best_state['current_position']][pickup_location]
                    reasoning.append(template.format(distance=distance_to_pickup))
                elif '{alternatives_count}' in template:
                    reasoning.append(template.format(alternatives_count=len(alternatives) - 1))
                else:
                    reasoning.append(template.format(time_window_status='Compatible'))
            
            # Calculate key factors
            distance_factor = self.instance.distances[best_state['current_position']][pickup_location]
            load_factor = best_state['load'] / self.instance.vehicle_capacity
            capacity_factor = (best_state['load'] + demand) / self.instance.vehicle_capacity
            
            key_factors = {
                'distance_proximity': 1.0 / (1.0 + distance_factor),
                'load_efficiency': 1.0 - load_factor,
                'capacity_utilization': capacity_factor,
                'alternative_options': len(alternatives) / len(current_vehicle_states)
            }
            
            confidence = decision['confidence']
            
            explanation = DecisionExplanation(
                decision_id=decision_id,
                decision_type='route_assignment',
                reasoning=reasoning,
                key_factors=key_factors,
                alternatives=alternatives[:3],  # Top 3 alternatives
                confidence=confidence,
                timestamp=time.time()
            )
        else:
            decision = {
                'action': 'no_assignment',
                'reason': 'no_suitable_vehicle',
                'confidence': 0.8
            }
            
            explanation = DecisionExplanation(
                decision_id=decision_id,
                decision_type='route_assignment',
                reasoning=[
                    f"Cannot assign request {request_id}",
                    f"Demand: {demand}",
                    f"No vehicle has sufficient capacity"
                ],
                key_factors={'capacity_constraint': 1.0},
                alternatives=[],
                confidence=0.8,
                timestamp=time.time()
            )
        
        return decision, explanation
    
    def _adjust_priority_with_explanation(self, context: Dict[str, Any], decision_id: str) -> Tuple[Dict[str, Any], DecisionExplanation]:
        """Adjust priority with explanation"""
        
        request_id = context['request_id']
        old_priority = context['current_priority']
        
        # Simple priority adjustment logic
        urgency_factor = context.get('urgency', 0.5)
        customer_importance = context.get('customer_importance', 0.5)
        time_pressure = context.get('time_pressure', 0.5)
        
        priority_score = (urgency_factor * 0.4 + customer_importance * 0.3 + time_pressure * 0.3)
        
        if priority_score > 0.7:
            new_priority = 'high'
            adjustment_reason = 'High urgency and customer importance'
            schedule_impact = 'Expedited processing required'
        elif priority_score > 0.4:
            new_priority = 'medium'
            adjustment_reason = 'Normal processing requirements'
            schedule_impact = 'Standard processing timeline'
        else:
            new_priority = 'low'
            adjustment_reason = 'Low urgency, flexible timeline'
            schedule_impact = 'Can be processed during off-peak hours'
        
        decision = {
            'action': 'priority_adjustment',
            'request_id': request_id,
            'old_priority': old_priority,
            'new_priority': new_priority,
            'confidence': 0.8
        }
        
        reasoning = []
        for template in self.explanation_templates['priority_adjustment']:
            reasoning.append(template.format(
                request_id=request_id,
                old_priority=old_priority,
                new_priority=new_priority,
                adjustment_reason=adjustment_reason,
                schedule_impact=schedule_impact
            ))
        
        key_factors = {
            'urgency_factor': urgency_factor,
            'customer_importance': customer_importance,
            'time_pressure': time_pressure,
            'priority_score': priority_score
        }
        
        explanation = DecisionExplanation(
            decision_id=decision_id,
            decision_type='priority_adjustment',
            reasoning=reasoning,
            key_factors=key_factors,
            alternatives=[],
            confidence=0.8,
            timestamp=time.time()
        )
        
        return decision, explanation
    
    def _optimize_route_with_explanation(self, context: Dict[str, Any], decision_id: str) -> Tuple[Dict[str, Any], DecisionExplanation]:
        """Optimize route with explanation"""
        
        vehicle_id = context['vehicle_id']
        current_route = context['current_route']
        
        current_length = self.calculate_route_distance(current_route)
        
        # Simple 2-opt optimization
        improved_route = current_route.copy()
        best_length = current_length
        
        for i in range(1, len(current_route) - 2):
            for j in range(i + 1, len(current_route) - 1):
                # Try 2-opt swap
                new_route = current_route[:i] + current_route[i:j][::-1] + current_route[j:]
                new_length = self.calculate_route_distance(new_route)
                
                if new_length < best_length:
                    improved_route = new_route
                    best_length = new_length
        
        potential_savings = current_length - best_length
        
        if potential_savings > 0.1:  # Only if meaningful improvement
            decision = {
                'action': 'route_optimization',
                'vehicle_id': vehicle_id,
                'old_route': current_route,
                'new_route': improved_route,
                'distance_saved': potential_savings,
                'confidence': 0.9
            }
            
            reasoning = []
            for template in self.explanation_templates['route_optimization']:
                reasoning.append(template.format(
                    vehicle_id=vehicle_id,
                    current_length=current_length,
                    potential_savings=potential_savings,
                    constraints='Capacity, Time windows, Pickup-delivery precedence',
                    method='2-opt local search'
                ))
            
            key_factors = {
                'route_complexity': len(current_route) / 10.0,
                'improvement_potential': potential_savings / current_length,
                'optimization_feasibility': 0.8,
                'computational_effort': len(current_route) ** 2
            }
        else:
            decision = {
                'action': 'no_optimization',
                'vehicle_id': vehicle_id,
                'reason': 'route_already_optimal',
                'confidence': 0.8
            }
            
            reasoning = [
                f"Route for vehicle {vehicle_id} is already optimal",
                f"Current route length: {current_length:.1f} km",
                f"No meaningful improvements found"
            ]
            
            key_factors = {
                'route_optimality': 1.0,
                'search_exhaustiveness': 1.0,
                'current_efficiency': 1.0
            }
        
        explanation = DecisionExplanation(
            decision_id=decision_id,
            decision_type='route_optimization',
            reasoning=reasoning,
            key_factors=key_factors,
            alternatives=[],
            confidence=decision['confidence'],
            timestamp=time.time()
        )
        
        return decision, explanation

print("Explainable AI class defined")
```

```python
class HumanOperator:
    """Human operator in the human-AI partnership"""
    
    def __init__(self, operator_id: str, expertise_level: float = 0.7):
        self.operator_id = operator_id
        self.expertise_level = expertise_level  # 0-1 scale
        self.trust_metrics = TrustMetrics(
            human_id=operator_id,
            trust_score=0.5,  # Initial neutral trust
            competence_trust=0.5,
            integrity_trust=0.6,
            benevolence_trust=0.5,
            interaction_count=0,
            satisfaction_history=[]
        )
        self.feedback_history = []
        self.intervention_history = []
        self.preferences = {
            'detail_preference': 0.7,  # How much detail they want in explanations
            'autonomy_preference': 0.6,  # How much AI autonomy they prefer
            'intervention_threshold': 0.3  # When they prefer to intervene
        }
    
    def evaluate_explanation(self, explanation: DecisionExplanation) -> Dict[str, Any]:
        """Evaluate AI explanation and provide feedback"""
        
        # Evaluate explanation quality based on operator expertise and preferences
        reasoning_quality = min(1.0, len(explanation.reasoning) / 5.0)  # More reasoning = better
        factor_transparency = len(explanation.key_factors) / 4.0  # More factors = more transparent
        alternative_awareness = len(explanation.alternatives) / 3.0  # More alternatives = more thorough
        
        # Weighted evaluation based on preferences
        explanation_score = (
            reasoning_quality * 0.4 +
            factor_transparency * 0.3 +
            alternative_awareness * 0.3
        )
        
        # Adjust for expertise level
        if self.expertise_level > 0.7:
            # Expert operators want more technical detail
            explanation_score *= (1 + 0.1 * (self.expertise_level - 0.7))
        
        # Determine satisfaction
        satisfaction = explanation_score
        
        # Update trust based on explanation quality
        trust_impact = (explanation_score - 0.5) * 0.1  # Small incremental changes
        
        evaluation = {
            'explanation_quality': explanation_score,
            'satisfaction': satisfaction,
            'trust_impact': trust_impact,
            'understanding_level': min(1.0, explanation_score * (1 + self.expertise_level * 0.2)),
            'feedback_needed': explanation_score < 0.6 or explanation.confidence < 0.7
        }
        
        return evaluation
    
    def provide_feedback(self, decision_id: str, evaluation: Dict[str, Any], comments: str = "") -> HumanFeedback:
        """Provide feedback on AI decision"""
        
        feedback_id = f"feedback_{len(self.feedback_history)}_{int(time.time())}"
        
        # Convert satisfaction to 1-5 rating
        rating = max(1, min(5, int(evaluation['satisfaction'] * 5)))
        
        feedback = HumanFeedback(
            feedback_id=feedback_id,
            decision_id=decision_id,
            human_id=self.operator_id,
            rating=rating,
            comments=comments,
            corrections={},  # Could be expanded with actual corrections
            trust_impact=evaluation['trust_impact'],
            timestamp=time.time()
        )
        
        self.feedback_history.append(feedback)
        
        # Update trust metrics
        self.trust_metrics.trust_score += evaluation['trust_impact']
        self.trust_metrics.trust_score = max(0, min(1, self.trust_metrics.trust_score))
        self.trust_metrics.interaction_count += 1
        self.trust_metrics.satisfaction_history.append(evaluation['satisfaction'])
        
        return feedback
    
    def should_intervene(self, decision: Dict[str, Any], explanation: DecisionExplanation) -> bool:
        """Decide whether to intervene in AI decision"""
        
        # Factors influencing intervention decision
        confidence_factor = 1.0 - explanation.confidence  # Lower confidence = more likely to intervene
        trust_factor = 1.0 - self.trust_metrics.trust_score  # Lower trust = more likely to intervene
        complexity_factor = len(explanation.reasoning) / 10.0  # More complex = more likely to intervene
        
        # Calculate intervention probability
        intervention_probability = (
            confidence_factor * 0.4 +
            trust_factor * 0.3 +
            complexity_factor * 0.2 +
            (1.0 - self.preferences['autonomy_preference']) * 0.1
        )
        
        # Adjust for expertise
        if self.expertise_level > 0.8:
            intervention_probability *= 1.2  # Experts more likely to intervene
        
        return random.random() < intervention_probability
    
    def make_intervention(self, decision: Dict[str, Any], explanation: DecisionExplanation) -> Dict[str, Any]:
        """Make human intervention in AI decision"""
        
        intervention = {
            'intervention_id': f"intervention_{len(self.intervention_history)}_{int(time.time())}",
            'human_id': self.operator_id,
            'original_decision': decision,
            'original_explanation': explanation,
            'intervention_type': 'override',  # Could be 'modify', 'reject', 'approve_with_conditions'
            'modified_decision': None,
            'reasoning': f"Human {self.operator_id} intervention based on expertise level {self.expertise_level:.1f}",
            'timestamp': time.time()
        }
        
        # Simple intervention logic (in practice would be more sophisticated)
        if decision.get('action') == 'assign_request':
            # Human might choose a different vehicle
            if explanation.alternatives and len(explanation.alternatives) > 1:
                # Choose second-best alternative
                alt_vehicle = explanation.alternatives[1]['vehicle_id']
                intervention['modified_decision'] = decision.copy()
                intervention['modified_decision']['vehicle_id'] = alt_vehicle
                intervention['reasoning'] += f" - Selected alternative vehicle {alt_vehicle} for better load balancing"
        
        self.intervention_history.append(intervention)
        
        return intervention

print("Human Operator class defined")
```

```python
try:
    class HumanAIPartnership:
        """Human-AI Symbiotic Partnership system"""
        
        def __init__(self, instance: VRPPDInstance):
            self.instance = instance
            self.ai_system = ExplainableAI(instance)
            self.human_operators = {}
            self.partnership_history = []
            self.performance_metrics = {
                'total_decisions': 0,
                'human_interventions': 0,
                'intervention_rate': 0.0,
                'average_trust': 0.0,
                'decision_quality_score': 0.0,
                'partnership_satisfaction': 0.0
            }
        
        def add_human_operator(self, operator_id: str, expertise_level: float = 0.7):
            """Add a human operator to the partnership"""
            self.human_operators[operator_id] = HumanOperator(operator_id, expertise_level)
        
        def make_collaborative_decision(self, decision_type: str, context: Dict[str, Any], 
                                       primary_operator: str = None) -> Dict[str, Any]:
            """Make collaborative decision with human-AI interaction"""
            
            # AI makes initial decision with explanation
            ai_decision, explanation = self.ai_system.make_decision_with_explanation(decision_type, context)
            
            # Select primary operator if not specified
            if primary_operator is None and self.human_operators:
                primary_operator = list(self.human_operators.keys())[0]
            
            collaborative_result = {
                'decision_id': explanation.decision_id,
                'decision_type': decision_type,
                'ai_decision': ai_decision,
                'explanation': explanation,
                'final_decision': ai_decision,
                'human_intervention': False,
                'operator_feedback': None,
                'partnership_outcome': 'ai_autonomous'
                'timestamp': time.time()
            }
            
            # Human evaluation and potential intervention
            if primary_operator and primary_operator in self.human_operators:
                operator = self.human_operators[primary_operator]
                
                # Human evaluates explanation
                evaluation = operator.evaluate_explanation(explanation)
                
                # Check if human should intervene
                if operator.should_intervene(ai_decision, explanation):
                    intervention = operator.make_intervention(ai_decision, explanation)
                    collaborative_result['human_intervention'] = True
                    collaborative_result['intervention'] = intervention
                    collaborative_result['final_decision'] = intervention.get('modified_decision', ai_decision)
                    collaborative_result['partnership_outcome'] = 'human_intervention'
                    
                    # Update performance metrics
                    self.performance_metrics['human_interventions'] += 1
                
                # Human provides feedback
                feedback = operator.provide_feedback(
                    explanation.decision_id, 
                    evaluation, 
                    f"Decision quality: {evaluation['satisfaction']:.2f}, Understanding: {evaluation['understanding_level']:.2f}"
                )
                
                collaborative_result['operator_feedback'] = feedback
                collaborative_result['evaluation'] = evaluation
            
            # Update performance metrics
            self.performance_metrics['total_decisions'] += 1
            self.performance_metrics['intervention_rate'] = (
                self.performance_metrics['human_interventions'] / self.performance_metrics['total_decisions']
            )
            
            # Update average trust
            if self.human_operators:
                avg_trust = np.mean([op.trust_metrics.trust_score for op in self.human_operators.values()])
                self.performance_metrics['average_trust'] = avg_trust
            
            # Store partnership history
            self.partnership_history.append(collaborative_result)
            
            return collaborative_result
        
        def simulate_partnership_session(self, num_decisions: int = 50) -> List[Dict[str, Any]]:
            """Simulate a partnership session with multiple decisions"""
            
            print(f"Starting Human-AI Partnership Session with {num_decisions} decisions...")
            
            session_results = []
            
            # Initialize vehicle states
            vehicle_states = {}
            for vehicle_id in self.instance.vehicles:
                vehicle_states[vehicle_id] = {
                    'current_position': self.instance.depot_start,
                    'load': 0,
                    'route': [self.instance.depot_start]
                }
            
            for decision_num in range(num_decisions):
                # Generate random decision context
                decision_type = random.choice(['route_assignment', 'priority_adjustment', 'route_optimization'])
                
                if decision_type == 'route_assignment':
                    # Generate pickup-delivery request
                    pickup_idx = random.choice(self.instance.pickups)
                    delivery_idx = pickup_idx + self.instance.num_pairs
                    demand = random.randint(1, self.instance.vehicle_capacity // 2)
                    
                    context = {
                        'request_id': f"req_{decision_num}",
                        'pickup_location': pickup_idx,
                        'delivery_location': delivery_idx,
                        'demand': demand,
                        'vehicle_states': vehicle_states
                    }
                
                elif decision_type == 'priority_adjustment':
                    context = {
                        'request_id': f"req_{decision_num}",
                        'current_priority': random.choice(['low', 'medium', 'high']),
                        'urgency': random.uniform(0, 1),
                        'customer_importance': random.uniform(0, 1),
                        'time_pressure': random.uniform(0, 1)
                    }
                
                else:  # route_optimization
                    vehicle_id = random.choice(self.instance.vehicles)
                    current_route = vehicle_states[vehicle_id]['route']
                    
                    # Add some random stops if route is too short
                    if len(current_route) < 3:
                        for _ in range(random.randint(1, 3)):
                            current_route.append(random.choice(self.instance.pickups + self.instance.deliveries))
                    
                    context = {
                        'vehicle_id': vehicle_id,
                        'current_route': current_route
                    }
                
                # Make collaborative decision
                result = self.make_collaborative_decision(decision_type, context)
                session_results.append(result)
                
                # Update vehicle states based on decision
                if decision_type == 'route_assignment' and result['final_decision'].get('action') == 'assign_request':
                    vehicle_id = result['final_decision']['vehicle_id']
                    vehicle_states[vehicle_id]['load'] += context['demand']
                    vehicle_states[vehicle_id]['route'].extend([context['pickup_location'], context['delivery_location']])
                
                elif decision_type == 'route_optimization' and result['final_decision'].get('action') == 'route_optimization':
                    vehicle_id = result['final_decision']['vehicle_id']
                    vehicle_states[vehicle_id]['route'] = result['final_decision']['new_route']
                
                # Report progress
                if (decision_num + 1) % 10 == 0:
                    print(f"  Decision {decision_num + 1}: {result['partnership_outcome']}, "
                          f"Trust: {self.performance_metrics['average_trust']:.2f}, "
                          f"Intervention Rate: {self.performance_metrics['intervention_rate']:.2f}")
            
            # Calculate final partnership metrics
            self._calculate_partnership_metrics()
            
            print(f"Partnership session completed!")
            return session_results
        
        def _calculate_partnership_metrics(self):
            """Calculate comprehensive partnership metrics"""
            
            if not self.partnership_history:
                return
            
            # Decision quality score
            ai_autonomous_decisions = [r for r in self.partnership_history if not r['human_intervention']]
            human_intervention_decisions = [r for r in self.partnership_history if r['human_intervention']]
            
            # Average explanation quality
            explanation_qualities = []
            for result in self.partnership_history:
                if 'evaluation' in result:
                    explanation_qualities.append(result['evaluation']['explanation_quality'])
            
            avg_explanation_quality = np.mean(explanation_qualities) if explanation_qualities else 0.5
            
            # Partnership satisfaction (combination of trust and explanation quality)
            partnership_satisfaction = (self.performance_metrics['average_trust'] + avg_explanation_quality) / 2
            
            # Decision improvement through collaboration
            collaboration_improvement = 0.1 * (len(human_intervention_decisions) / max(1, len(self.partnership_history)))
            
            decision_quality_score = (avg_explanation_quality + collaboration_improvement) * 100
            
            self.performance_metrics.update({
                'decision_quality_score': decision_quality_score,
                'partnership_satisfaction': partnership_satisfaction,
                'avg_explanation_quality': avg_explanation_quality,
                'ai_autonomous_rate': len(ai_autonomous_decisions) / len(self.partnership_history),
                'collaboration_improvement': collaboration_improvement
            })
    
    print("Human-AI Partnership class defined")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    def analyze_partnership_results(results: List[Dict[str, Any]]) -> Dict[str, Any]:
        """Analyze human-AI partnership performance"""
        
        if not results:
            return {'error': 'No results to analyze'}
        
        # Extract key metrics
        total_decisions = len(results)
        human_interventions = sum(1 for r in results if r['human_intervention'])
        intervention_rate = human_interventions / total_decisions
        
        # Analyze by decision type
        decision_types = {}
        for result in results:
            dt = result['decision_type']
            if dt not in decision_types:
                decision_types[dt] = {'total': 0, 'interventions': 0}
            decision_types[dt]['total'] += 1
            if result['human_intervention']:
                decision_types[dt]['interventions'] += 1
        
        # Analyze explanation quality
        explanation_qualities = []
        trust_scores = []
        satisfaction_scores = []
        
        for result in results:
            if 'evaluation' in result:
                explanation_qualities.append(result['evaluation']['explanation_quality'])
                satisfaction_scores.append(result['evaluation']['satisfaction'])
            
            if 'operator_feedback' in result:
                # Get operator trust score
                operator_id = result['operator_feedback'].human_id
                if operator_id in partnership.human_operators:
                    trust_scores.append(partnership.human_operators[operator_id].trust_metrics.trust_score)
        
        # Analyze trust development over time
        trust_development = []
        for operator_id, operator in partnership.human_operators.items():
            trust_development.append({
                'operator_id': operator_id,
                'initial_trust': 0.5,
                'final_trust': operator.trust_metrics.trust_score,
                'trust_improvement': operator.trust_metrics.trust_score - 0.5,
                'interactions': operator.trust_metrics.interaction_count,
                'avg_satisfaction': np.mean(operator.trust_metrics.satisfaction_history) if operator.trust_metrics.satisfaction_history else 0.5
            })
        
        return {
            'total_decisions': total_decisions,
            'human_interventions': human_interventions,
            'intervention_rate': intervention_rate,
            'decision_types': decision_types,
            'avg_explanation_quality': np.mean(explanation_qualities) if explanation_qualities else 0.5,
            'avg_trust_score': np.mean(trust_scores) if trust_scores else 0.5,
            'avg_satisfaction': np.mean(satisfaction_scores) if satisfaction_scores else 0.5,
            'trust_development': trust_development,
            'partnership_effectiveness': (1.0 - intervention_rate) * np.mean(explanation_qualities) if explanation_qualities else 0.5
        }
    
    # Analyze the partnership results
    analysis = analyze_partnership_results(session_results)
    
    print("\n=== HUMAN-AI PARTNERSHIP ANALYSIS ===")
    print(f"Total decisions: {analysis['total_decisions']}")
    print(f"Human interventions: {analysis['human_interventions']}")
    print(f"Intervention rate: {analysis['intervention_rate']:.2%}")
    print(f"Average explanation quality: {analysis['avg_explanation_quality']:.2f}")
    print(f"Average trust score: {analysis['avg_trust_score']:.2f}")
    print(f"Average satisfaction: {analysis['avg_satisfaction']:.2f}")
    print(f"Partnership effectiveness: {analysis['partnership_effectiveness']:.2f}")
    
    print("\nTrust Development by Operator:")
    for trust_info in analysis['trust_development']:
        print(f"- {trust_info['operator_id']}: {trust_info['initial_trust']:.2f} → {trust_info['final_trust']:.2f} "
              f"(+{trust_info['trust_improvement']:.2f}) in {trust_info['interactions']} interactions")
    
    print("\nIntervention Rate by Decision Type:")
    for decision_type, stats in analysis['decision_types'].items():
        rate = stats['interventions'] / stats['total'] if stats['total'] > 0 else 0
        print(f"- {decision_type}: {rate:.2%} ({stats['interventions']}/{stats['total']})")
except Exception as e:
    print(f'Error in cell: {e}')
```
