```python
# Import required packages for Ethical Framework
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from math import sqrt
import random
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(76)
random.seed(76)

print("Libraries imported successfully for Ethical Framework implementation")
```

```python
# Define problem data structures (same as previous tiers)
class SDVRPInstance:
    """Class to represent SDVRP problem instance"""
    def __init__(self, depot_coords, customer_coords, demands, vehicle_capacities):
        self.depot = depot_coords
        self.customers = customer_coords
        self.demands = demands
        self.capacities = vehicle_capacities
        self.n_customers = len(customer_coords)
        self.n_vehicles = len(vehicle_capacities)
        
    def calculate_distance_matrix(self):
        """Calculate Euclidean distance matrix between all nodes"""
        all_nodes = [self.depot] + self.customers
        n_nodes = len(all_nodes)
        distances = np.zeros((n_nodes, n_nodes))
        
        for i in range(n_nodes):
            for j in range(n_nodes):
                if i != j:
                    x1, y1 = all_nodes[i]
                    x2, y2 = all_nodes[j]
                    distances[i][j] = sqrt((x2 - x1)**2 + (y2 - y1)**2)
        
        return distances

# Create the same instance as previous tiers
depot_coords = (0, 0)
customer_coords = [(10, 15), (20, 5), (15, 20), (25, 10)]
demands = [70, 130, 60, 80]  # Customer demands
vehicle_capacities = [100, 100]  # Two vehicles with capacity 100

instance = SDVRPInstance(depot_coords, customer_coords, demands, vehicle_capacities)
distance_matrix = instance.calculate_distance_matrix()

print(f"SDVRP Instance for Ethical Framework:")
print(f"- Depot: {depot_coords}")
print(f"- Customers: {len(customer_coords)} with demands: {demands}")
print(f"- Vehicles: {len(vehicle_capacities)} with capacities: {vehicle_capacities}")
print(f"- Total demand: {sum(demands)} units")
print(f"- Total capacity: {sum(vehicle_capacities)} units")
```

```python
# Stakeholder identification and utility functions
class StakeholderAnalysis:
    """Analyze stakeholder interests and utility functions"""
    
    def __init__(self, instance):
        self.instance = instance
        self.stakeholders = {
            'customers': {
                'priority': 0.3,  # Service level priority
                'cost_sensitivity': 0.2,  # Price sensitivity
                'accessibility': 0.25, # Geographic accessibility
                'loyalty': 0.15, # Customer loyalty
            },
            'employees': {
                'workload_balance': 0.4, # Workload balance preference
                'safety': 0.5, # Safety priority
                'job_satisfaction': 0.3, # Job satisfaction
                'training_needs': 0.2  # Training and development
            },
            'environment': {
                'emissions': 0.3,  # Environmental impact
                'noise_pollution': 0.2,  # Noise pollution
                'energy_efficiency': 0.4, # Energy efficiency
                'sustainability': 0.3  # Long-term sustainability
            },
            'company': {
                'profitability': 0.6,  # Profit maximization
                'market_share': 0.3,  # Market position
                'reputation': 0.2,  # Company reputation
                'growth': 0.1   # Growth potential
            },
            'regulators': {
                'compliance': 0.7,  # Regulatory compliance
                'safety_standards': 0.8,  # Safety standards
                'public_interest': 0.6,  # Public interest
                'transparency': 0.5   # Transparency requirements
            }
        }
        
    def calculate_stakeholder_utility(self, solution, stakeholder, metrics):
        """Calculate utility for a specific stakeholder"""
        
        utility = 0.0
        
        if stakeholder == 'customers':
            # Service level utility
            service_level = self._calculate_service_level(solution, metrics)
            utility += stakeholder['priority'] * service_level
            
            # Cost sensitivity
            avg_cost_per_unit = metrics['total_cost'] / metrics['total_delivered']
            cost_utility = 1 / (1 + avg_cost_per_unit * stakeholder['cost_sensitivity'])
            utility += stakeholder['cost_sensitivity'] * cost_utility
            
            # Accessibility utility
            avg_distance = metrics['avg_distance_per_delivery']
            accessibility_utility = 1 / (1 + avg_distance / 50 * stakeholder['accessibility'])
            utility += stakeholder['accessibility'] * accessibility_utility
            
            # Loyalty utility
            on_time_delivery = metrics['on_time_delivery_rate']
            loyalty_utility = on_time_delivery * stakeholder['loyalty']
            utility += stakeholder['loyalty'] * loyalty_utility
            
        elif stakeholder == 'employees':
            # Workload balance utility
            workload_variance = metrics['workload_variance']
            workload_utility = 1 / (1 + workload_variance * stakeholder['workload_balance'])
            utility += stakeholder['workload_balance'] * workload_utility
            
            # Safety utility
            safety_score = metrics['safety_compliance']
            safety_utility = safety_score * stakeholder['safety']
            utility += stakeholder['safety'] * safety_utility
            
            # Job satisfaction utility
            satisfaction_score = metrics['job_satisfaction']
            satisfaction_utility = satisfaction_score * stakeholder['job_satisfaction']
            utility += stakeholder['job_satisfaction'] * satisfaction_utility
            
        elif stakeholder == 'environment':
            # Environmental impact utility
            emissions = metrics['total_emissions']
            emissions_utility = 1 / (1 + emissions / 100 * stakeholder['emissions'])
            utility += stakeholder['emissions'] * emissions_utility
            
            # Noise pollution
            noise_level = metrics['noise_pollution']
            noise_utility = 1 / (1 + noise_level / 50 * stakeholder['noise_pollution'])
            utility += stakeholder['noise_pollution'] * noise_utility
            
            # Energy efficiency
            fuel_consumption = metrics['fuel_consumption']
            energy_utility = 1 / (1 + fuel_consumption / 1000 * stakeholder['energy_efficiency'])
            utility += stakeholder['energy_efficiency'] * energy_utility
            
        elif stakeholder == 'company':
            # Profitability utility
            profit_margin = metrics['profit_margin']
            profitability_utility = profit_margin * stakeholder['profitability']
            utility += stakeholder['profitability'] * profitability_utility
            
            # Market share utility
            market_position = metrics['market_share']
            market_utility = market_position * stakeholder['market_share']
            utility += stakeholder['market_share'] * market_utility
            
        elif stakeholder == 'regulators':
            # Compliance utility
            compliance_score = metrics['regulatory_compliance']
            compliance_utility = compliance_score * stakeholder['compliance']
            utility += stakeholder['compliance'] * compliance_utility
            
        return utility
    
    def _calculate_service_level(self, solution, metrics):
        """Calculate service level from solution"""
        
        total_delivered = metrics['total_delivered']
        total_demand = sum(self.instance.demands)
        
        if total_demand > 0:
            service_level = total_delivered / total_demand
        else:
            service_level = 0.0
        
        return service_level

print("Stakeholder Analysis class defined successfully")
```

```python
# Ethical Constraints and Compliance Framework
class EthicalFramework:
    """Ethical framework for SDVRP decision making"""
    
    def __init__(self, instance, stakeholder_analysis):
        self.instance = instance
        self.stakeholder_analysis = stakeholder_analysis
        self.constraints = []
        self.compliance_score = 0.0
        
        # Define ethical principles
        self.principles = {
            'fairness': {
                'description': 'Ensure equitable service across all customer segments',
                'weight': 0.25,
                'implementation': self._fairness_constraint
            },
            'sustainability': {
                'description': 'Minimize environmental impact and resource usage',
                'weight': 0.20,
                'implementation': self._sustainability_constraint
            },
            'transparency': {
                'description': 'Provide clear reasoning for routing decisions',
                'weight': 0.15,
                'implementation': self._transparency_constraint
            },
            'accountability': {
                'description': 'Maintain audit trail for all decisions',
                'weight': 0.15,
                'implementation': self._accountability_constraint
            },
        }
        
    def _fairness_constraint(self, solution, metrics):
        """Ensure fair service distribution"""
        
        # Calculate service level variance
        service_levels = [self._calculate_customer_service_level(solution, i+1) for i in range(self.instance.n_customers)]
        
        if len(service_levels) > 1:
            service_variance = np.var(service_levels)
            fairness_score = max(0, 1 - service_variance)
        else:
            fairness_score = 1.0  # All customers get same service
        
        return fairness_score
    
    def _sustainability_constraint(self, solution, metrics):
        """Minimize environmental impact"""
        
        # Calculate environmental metrics
        total_distance = metrics['total_distance']
        emissions_per_km = 0.5  # kg CO2 per km
        total_emissions = total_distance * emissions_per_km
        
        # Sustainability score (inverse of emissions)
        sustainability_score = max(0, 1 - total_emissions / 1000)  # Normalize to 0-1 scale
        
        return sustainability_score
    
    def _transparency_constraint(self, solution, metrics):
        """Ensure transparent decision making"""
        
        # Calculate transparency score based on route complexity
        route_complexity = len(solution['routes'])
        documentation_quality = 0.8  # Assume 80% documentation quality
        
        transparency_score = (route_complexity * documentation_quality) / 2.0
        
        return transparency_score
    
    def _accountability_constraint(self, solution, metrics):
        """Maintain audit trail"""
        
        # Calculate auditability score
        decision_traceability = 0.9  # Assume good traceability
        
        return decision_traceability
    
    def evaluate_solution(self, solution, metrics):
        """Evaluate solution against ethical constraints"""
        
        # Calculate ethical compliance scores
        fairness_score = self._fairness_constraint(solution, metrics)
        sustainability_score = self._sustainability_constraint(solution, metrics)
        transparency_score = self._transparency_constraint(solution, metrics)
        accountability_score = self._accountability_constraint(solution, metrics)
        
        # Overall ethical score
        ethical_score = (fairness_score * 0.25 + 
                           sustainability_score * 0.20 + 
                           transparency_score * 0.15 + 
                           accountability_score * 0.15)
        
        return ethical_score
    
    def add_constraint(self, principle, implementation_function):
        """Add a custom ethical constraint"""
        self.constraints.append({
            'principle': principle,
            'implementation': implementation_function,
        })
    
    def check_compliance(self, solution, metrics):
        """Check if solution meets all ethical constraints"""
        
        for constraint in self.constraints:
            compliance_score = constraint['implementation'](solution, metrics)
            if compliance_score < 0.7:
                return False, f"Constraint violation: {constraint['principle']}"
        
        return True

print("Ethical Framework class defined successfully")
```

```python
# Value-Aligned Optimization Engine
class ValueAlignedOptimizer:
    """Multi-objective optimizer with ethical considerations"""
    
    def __init__(self, instance, stakeholder_analysis, ethical_framework):
        self.instance = instance
        self.stakeholder_analysis = stakeholder_analysis
        self.ethical_framework = ethical_framework
        self.distance_matrix = instance.calculate_distance_matrix()
        
    def optimize_with_ethics(self, base_solution, metrics):
        """Optimize solution while maintaining ethical standards"""
        
        print("=== VALUE-ALIGNED OPTIMIZATION ===")
        
        # Calculate base metrics
        base_metrics = {
            'total_distance': base_solution['total_distance'],
            'total_delivered': sum(sum(d.values()) for d in base_solution['deliveries']),
            'vehicles_used': len(base_solution['routes'])
        }
        
        print(f"Base solution: Distance = {base_metrics['total_distance']:.2f}")
        print(f"Base deliveries: {base_metrics['total_delivered']}")
        
        # Generate alternative solutions with ethical considerations
        ethical_solutions = self._generate_ethical_alternatives(base_solution, metrics)
        
        # Evaluate all solutions
        best_solution = base_solution
        best_ethical_score = self.ethical_framework.evaluate_solution(base_solution, metrics)
        best_total_utility = 0
        
        for i, (solution, ethical_score) in enumerate(ethical_solutions):
            # Calculate stakeholder utility for this solution
            total_utility = 0
            
            for stakeholder in self.stakeholder_analysis.stakeholders:
                utility = self.stakeholder_analysis.calculate_stakeholder_utility(solution, stakeholder, metrics)
                total_utility += utility
            
            # Combined utility with ethical score
            combined_score = total_utility * (0.7 + 0.3 * ethical_score)
            
            if combined_score > best_total_utility:
                best_solution = solution
                best_total_utility = combined_score
                best_ethical_score = ethical_score
        
        print(f"\nOptimized solution:")
        print(f"  Distance: {best_solution['total_distance']:.2f}")
        print(f"  Ethical score: {best_ethical_score:.2f}")
        print(f"  Stakeholder utility: {best_total_utility:.2f}")
        
        return best_solution
    
    def _generate_ethical_alternatives(self, base_solution, metrics):
        """Generate alternative solutions with ethical considerations"""
        
        alternatives = []
        
        # Alternative 1: Prioritize high-priority customers
        alt1 = base_solution.copy()
        alt1['routes'] = [[0, 2, 4, 0], [0, 1, 3, 0]]  # Prioritize Customer 2 and 4
        alternatives.append(alt1)
        
        # Alternative 2: Minimize distance
        alt2 = base_solution.copy()
        # Simplified distance minimization
        alt2['routes'] = [[0, 1, 3, 0], [0, 2, 4, 0]]
        alternatives.append(alt2)
        
        # Alternative 3: Balance workload
        alt3 = base_solution.copy()
        # Balance workload between vehicles
        alt3['routes'] = [[0, 1, 0], [0, 2, 3, 0], [0, 4, 0]]
        alternatives.append(alt3)
        
        return alternatives

print("Value-Aligned Optimizer class defined successfully")
```

```python
# Regulatory Compliance and Risk Assessment
class ComplianceMonitor:
    """Monitor regulatory compliance and risk factors"""
    
    def __init__(self, instance):
        self.instance = instance
        self.distance_matrix = instance.calculate_distance_matrix()
        self.regulations = {
            'driving_hours': {'max_daily_hours': 8, 'min_rest_time': 0.5},
            'vehicle_maintenance': {'inspection_interval_days': 30, 'safety_checks': True},
            'emissions': {'max_co2_per_km': 0.5, 'annual_limit': 10000},
            'noise': {'max_db_level': 70, 'night_restrictions': True},
        }
        
        self.violations = []
        self.risk_assessment = {}
        
    def check_compliance(self, solution, metrics):
        """Check solution against regulatory requirements"""
        
        compliance_issues = []
        
        # Check driving hours
        total_distance = solution['total_distance']
        avg_speed = 50  # km/h
        estimated_hours = total_distance / avg_speed
        
        if estimated_hours > self.regulations['driving_hours']['max_daily_hours']:
            compliance_issues.append("Driving hours exceed maximum allowed")
        
        # Check emissions
        total_emissions = total_distance * self.regulations['emissions']['max_co2_per_km']
        if total_emissions > self.regulations['emissions']['annual_limit']:
            compliance_issues.append("Annual emissions exceed regulatory limit")
        
        # Check noise restrictions
        if self.regulations['noise']['night_restrictions']:
            compliance_issues.append("Noise restrictions violated during night hours")
        
        # Check maintenance requirements
        days_since_last_maintenance = 15  # Assume 15 days since last maintenance
        if days_since_last_maintenance > self.regulations['vehicle_maintenance']['inspection_interval_days']:
            compliance_issues.append("Vehicle maintenance overdue")
        
        self.violations.extend(compliance_issues)
        
        # Calculate compliance score
        max_violations = 10
        compliance_score = max(0, (max_violations - len(self.violations)) / max_violations)
        
        return compliance_score, compliance_issues
    
    def assess_risks(self, solution, metrics):
        """Assess operational risks"""
        
        risk_factors = {
            'route_complexity': len(solution['routes']),
            'vehicle_utilization': metrics['vehicle_utilization'],
            'demand_volatility': 0.3,  # Assumed demand volatility
            'weather_sensitivity': 0.2,  # Weather impact
            'traffic_conditions': 0.1,  # Traffic conditions
        }
        
        # Calculate risk scores
        risk_score = 1.0
        for factor, weight in risk_factors.items():
            risk_score *= (1 - factor * weight)
        
        return risk_score

print("Compliance Monitor class defined successfully")
```

```python
# Main Ethical Optimization Execution
def run_ethical_optimization(instance, distance_matrix):
    """Run complete ethical optimization"""
    
    print("=== ETHICAL OPTIMIZATION EXECUTION ===")
    
    # Initialize components
    stakeholder_analysis = StakeholderAnalysis(instance)
    ethical_framework = EthicalFramework(instance, stakeholder_analysis)
    optimizer = ValueAlignedOptimizer(instance, stakeholder_analysis, ethical_framework)
    compliance_monitor = ComplianceMonitor(instance)
    
    # Generate baseline solution (from previous tier)
    base_solution = {
        'routes': [[0, 1, 3, 0], [0, 2, 4, 0]],
        'deliveries': [{1: 70, 2: 60}, {3: 60, 4: 80}],
        'method': 'baseline',
        'total_distance': 85.0
    }
    
    print(f"Baseline solution: Distance = {base_solution['total_distance']:.2f}")
    
    # Calculate baseline metrics
    base_metrics = {
        'total_distance': base_solution['total_distance'],
        'total_delivered': sum(sum(d.values()) for d in base_solution['deliveries']),
        'vehicles_used': len(base_solution['routes']),
        'vehicle_utilization': 80.0,
    }
    
    # Run value-aligned optimization
    ethical_solution = optimizer.optimize_with_ethics(base_solution, base_metrics)
    
    # Check compliance and risks
    compliance_score, compliance_issues = compliance_monitor.check_compliance(ethical_solution, base_metrics)
    risk_score = compliance_monitor.assess_risks(ethical_solution, base_metrics)
    
    print(f"\n=== ETHICAL OPTIMIZATION RESULTS ===")
    print(f"Baseline distance: {base_metrics['total_distance']:.2f}")
    print(f"Ethical solution distance: {ethical_solution['total_distance']:.2f}")
    print(f"Distance change: {((ethical_solution['total_distance'] - base_metrics['total_distance']) / base_metrics['total_distance']) * 100:+.1f}%")
    
    print(f"Compliance score: {compliance_score:.2f}")
    print(f"Risk assessment: {risk_score:.2f}")
    
    if compliance_issues:
        print(f"Compliance issues: {len(compliance_issues)}")
        for issue in compliance_issues:
            print(f"  - {issue}")
    
    return {
        'base_solution': base_solution,
        'ethical_solution': ethical_solution,
        'compliance_score': compliance_score,
        'risk_score': risk_score
    }

# Run ethical optimization
results = run_ethical_optimization(instance, distance_matrix)
```
