## Tier 8: The Value-Aligned & Ethical Framework

### Key assumptions
- Multi-stakeholder utility optimization balancing economic, social, and environmental objectives
- Constitutional AI with embedded ethical principles and value alignment constraints
- Stakeholder feedback integration with continuous ethical constraint learning
- Bias detection and mitigation systems ensuring fair and equitable routing decisions
- Transparency and accountability mechanisms with comprehensive audit trails
- Ethical scoring system evaluating routing solutions against multiple value dimensions

### Approach (step-by-step)
1. **Ethical Framework Design**: Define comprehensive ethical principles and value constraints
2. **Stakeholder Modeling**: Create multi-stakeholder utility functions with competing objectives
3. **Constitutional AI**: Implement AI system with embedded ethical constraints and oversight
4. **Bias Detection**: Develop systems to identify and mitigate algorithmic and data biases
5. **Value Alignment**: Optimize routing solutions balancing economic efficiency with ethical considerations
6. **Transparency Systems**: Create comprehensive audit trails and explainable ethical decisions
7. **Continuous Learning**: Implement feedback loops for ethical constraint adaptation and improvement

### What to look for in the results
- Multi-stakeholder utility optimization with balanced objective achievement
- Ethical constraint satisfaction and violation detection mechanisms
- Bias detection outcomes and mitigation effectiveness analysis
- Stakeholder satisfaction metrics across different value dimensions
- Transparency and accountability audit trail completeness
- Comparison between utilitarian and value-aligned optimization approaches

### Concrete example (from the source)
The source describes an ethical routing scenario:
- **Stakeholder Conflict**: Economic efficiency vs environmental impact vs worker welfare
- **Ethical Dilemma**: Route A offers 15% cost savings but 25% higher emissions and driver fatigue
- **Constitutional AI**: System applies ethical principles to evaluate stakeholder impacts
- **Value Alignment**: Route B selected with 8% higher cost but 40% lower emissions and better driver conditions
- **Stakeholder Feedback**: Customer satisfaction 92%, environmental compliance 100%, driver satisfaction 88%
- **Audit Trail**: Complete ethical decision documentation with constraint satisfaction analysis

### Why this Tier exists vs Tiers 1-7
This Tier provides ethical optimization capabilities:
- **Value Alignment**: Multi-stakeholder utility vs single-objective optimization
- **Ethical Constraints**: Constitutional AI vs unconstrained algorithmic optimization
- **Bias Mitigation**: Fair and equitable decisions vs potentially biased algorithms
- **Transparency**: Explainable ethical reasoning vs black-box optimization
- **Accountability**: Comprehensive audit trails vs untraceable decisions

### Pros / Cons vs Tiers 1-7
**Pros:**
- **Ethical Compliance**: Ensures routing decisions align with organizational values and regulations
- **Stakeholder Balance**: Optimizes for multiple stakeholders rather than single economic objectives
- **Risk Mitigation**: Reduces ethical, legal, and reputational risks through systematic constraint enforcement
- **Social Responsibility**: Demonstrates commitment to environmental sustainability and social welfare
- **Long-term Value**: Creates sustainable competitive advantage through ethical positioning
- **Regulatory Alignment**: Proactively addresses emerging regulations and compliance requirements

**Cons:**
- **Computational Complexity**: Multi-objective optimization significantly increases problem complexity
- **Solution Trade-offs**: Ethical constraints may reduce pure economic efficiency and increase costs
- **Implementation Challenges**: Requires sophisticated ethical modeling and stakeholder engagement
- **Measurement Difficulty**: Quantifying ethical objectives and stakeholder utilities can be challenging
- **Subjectivity Risks**: Ethical principles may vary across cultures and organizational contexts
- **Performance Overhead**: Ethical evaluation and constraint checking adds computational burden

### When to use this Tier
- **Regulated Industries**: When compliance with environmental, labor, and safety regulations is essential
- **Public Sector**: When serving diverse stakeholders with competing interests and values
- **Sustainability-Focused Organizations**: When environmental and social responsibility are strategic priorities
- **High-Profile Operations**: When public perception and brand reputation are critical success factors
- **Multi-Stakeholder Environments**: When balancing customer, employee, community, and shareholder interests
- **Emerging Markets**: When establishing ethical standards and competitive differentiation through responsible practices

```python
from typing import Tuple, List, Dict, Set

# Import required libraries for value-aligned and ethical framework
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from dataclasses import dataclass, field
from typing import List, Tuple, Dict, Optional, Any
import random
import time
from datetime import datetime, timedelta
from collections import defaultdict, deque
from enum import Enum
import warnings
warnings.filterwarnings('ignore')

# Set up visualization style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("✅ Libraries imported successfully for Value-Aligned & Ethical Framework")
```

```python
class EthicalPrinciple(Enum):
    """Core ethical principles for routing optimization"""
    ENVIRONMENTAL_SUSTAINABILITY = "environmental_sustainability"
    WORKER_WELFARE = "worker_welfare"
    CUSTOMER_SATISFACTION = "customer_satisfaction"
    ECONOMIC_EFFICIENCY = "economic_efficiency"
    SOCIAL_EQUITY = "social_equity"
    TRANSPARENCY = "transparency"
    ACCOUNTABILITY = "accountability"

@dataclass
class VRPBInstance:
    """Data class for VRP with Backhauls problem instance"""
    num_deliveries: int
    num_pickups: int
    delivery_demands: List[float]
    pickup_demands: List[float]
    capacity: float
    distance_matrix: np.ndarray
    
    def __post_init__(self):
        self.num_customers = self.num_deliveries + self.num_pickups
        self.delivery_indices = list(range(1, self.num_deliveries + 1))
        self.pickup_indices = list(range(self.num_deliveries + 1, 
                                         self.num_deliveries + self.num_pickups + 1))
        self.all_customer_indices = self.delivery_indices + self.pickup_indices

@dataclass
class Stakeholder:
    """Represents a stakeholder in the routing system"""
    stakeholder_id: str
    stakeholder_type: str
    priorities: Dict[EthicalPrinciple, float]
    utility_function: str
    weight: float
    feedback_history: List[Dict[str, Any]] = field(default_factory=list)

@dataclass
class EthicalConstraint:
    """Represents an ethical constraint in routing decisions"""
    constraint_id: str
    principle: EthicalPrinciple
    constraint_type: str
    threshold: float
    penalty_weight: float
    description: str
    satisfaction_level: float = 1.0

@dataclass
class EthicalAuditTrail:
    """Audit trail for ethical decision making"""
    audit_id: str
    timestamp: datetime
    decision_context: str
    considered_alternatives: List[Dict[str, Any]]
    ethical_evaluation: Dict[str, float]
    final_decision: Dict[str, Any]
    constraint_violations: List[str]
    stakeholder_impact: Dict[str, float]
    justification: str
    reviewer_id: str

@dataclass
class BiasDetection:
    """Bias detection and mitigation results"""
    bias_type: str
    detection_method: str
    bias_score: float
    affected_groups: List[str]
    mitigation_strategy: str
    mitigation_effectiveness: float

print("✅ Ethical framework data structures defined")
```

```python
def create_concrete_example() -> VRPBInstance:
    """Create the concrete example from the source material"""
    
    num_deliveries = 4
    num_pickups = 3
    delivery_demands = [5, 4, 6, 3]
    pickup_demands = [4, 5, 3]
    capacity = 20
    
    distance_matrix = np.array([
        [0, 3, 4, 5, 6, 8, 7, 9],
        [3, 0, 2, 3, 4, 6, 5, 7],
        [4, 2, 0, 2, 3, 5, 4, 6],
        [5, 3, 2, 0, 2, 4, 3, 5],
        [6, 4, 3, 2, 0, 3, 2, 4],
        [8, 6, 5, 4, 3, 0, 2, 3],
        [7, 5, 4, 3, 2, 2, 0, 2],
        [9, 7, 6, 5, 4, 3, 2, 0]
    ])
    
    return VRPBInstance(
        num_deliveries=num_deliveries,
        num_pickups=num_pickups,
        delivery_demands=delivery_demands,
        pickup_demands=pickup_demands,
        capacity=capacity,
        distance_matrix=distance_matrix
    )

instance = create_concrete_example()
print(f"✅ Created VRPB instance with {instance.num_deliveries} deliveries and {instance.num_pickups} pickups")
print(f"Vehicle capacity: {instance.capacity}")
print(f"Total delivery demand: {sum(instance.delivery_demands)}")
print(f"Total pickup demand: {sum(instance.pickup_demands)}")
```

```python
class ConstitutionalAI:
    """Constitutional AI system with embedded ethical principles"""
    
    def __init__(self, instance: VRPBInstance):
        self.instance = instance
        self.ethical_constraints = []
        self.stakeholders = {}
        self.audit_trail = []
        self.bias_detections = []
        self.constitution = self._initialize_constitution()
        
        self._initialize_stakeholders()
        self._initialize_ethical_constraints()
        
        print(f"🏛️ Constitutional AI initialized with {len(self.constitution)} principles")
    
    def _initialize_constitution(self) -> Dict[EthicalPrinciple, Dict[str, Any]]:
        """Initialize the ethical constitution"""
        constitution = {
            EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY: {
                "description": "Minimize environmental impact and carbon footprint",
                "weight": 0.25,
                "constraints": ["max_emissions", "min_fuel_efficiency"],
                "metrics": ["co2_emissions", "fuel_consumption", "noise_pollution"]
            },
            EthicalPrinciple.WORKER_WELFARE: {
                "description": "Ensure safe and fair working conditions for drivers",
                "weight": 0.20,
                "constraints": ["max_working_hours", "min_rest_periods", "fair_wage_distribution"],
                "metrics": ["driver_fatigue", "working_hours", "rest_time", "wage_fairness"]
            },
            EthicalPrinciple.CUSTOMER_SATISFACTION: {
                "description": "Maximize customer satisfaction and service quality",
                "weight": 0.20,
                "constraints": ["max_wait_time", "min_service_quality", "equitable_service"],
                "metrics": ["delivery_time", "service_quality", "customer_feedback"]
            },
            EthicalPrinciple.ECONOMIC_EFFICIENCY: {
                "description": "Maintain economic viability and cost efficiency",
                "weight": 0.15,
                "constraints": ["max_operating_cost", "min_profit_margin"],
                "metrics": ["operating_cost", "profit_margin", "resource_utilization"]
            },
            EthicalPrinciple.SOCIAL_EQUITY: {
                "description": "Ensure fair and equitable service distribution",
                "weight": 0.10,
                "constraints": ["equitable_access", "fair_pricing", "non_discrimination"],
                "metrics": ["service_equity", "price_fairness", "accessibility"]
            },
            EthicalPrinciple.TRANSPARENCY: {
                "description": "Maintain transparency in decision making and operations",
                "weight": 0.05,
                "constraints": ["full_disclosure", "explainable_decisions"],
                "metrics": ["decision_transparency", "information_disclosure"]
            },
            EthicalPrinciple.ACCOUNTABILITY: {
                "description": "Ensure accountability for decisions and impacts",
                "weight": 0.05,
                "constraints": ["traceable_decisions", "impact_assessment"],
                "metrics": ["decision_traceability", "impact_monitoring"]
            }
        }
        return constitution
    
    def _initialize_stakeholders(self):
        """Initialize stakeholders with their priorities and utility functions"""
        stakeholders_config = [
            {
                "stakeholder_id": "CUSTOMER",
                "stakeholder_type": "customer",
                "priorities": {
                    EthicalPrinciple.CUSTOMER_SATISFACTION: 0.4,
                    EthicalPrinciple.ECONOMIC_EFFICIENCY: 0.3,
                    EthicalPrinciple.SOCIAL_EQUITY: 0.2,
                    EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY: 0.1
                },
                "utility_function": "service_quality_cost_balance",
                "weight": 0.3
            },
            {
                "stakeholder_id": "DRIVER",
                "stakeholder_type": "employee",
                "priorities": {
                    EthicalPrinciple.WORKER_WELFARE: 0.5,
                    EthicalPrinciple.ECONOMIC_EFFICIENCY: 0.3,
                    EthicalPrinciple.SOCIAL_EQUITY: 0.2
                },
                "utility_function": "welfare_fairness_balance",
                "weight": 0.25
            },
            {
                "stakeholder_id": "ENVIRONMENT",
                "stakeholder_type": "society",
                "priorities": {
                    EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY: 0.6,
                    EthicalPrinciple.SOCIAL_EQUITY: 0.3,
                    EthicalPrinciple.TRANSPARENCY: 0.1
                },
                "utility_function": "environmental_impact_minimizer",
                "weight": 0.2
            },
            {
                "stakeholder_id": "COMPANY",
                "stakeholder_type": "business",
                "priorities": {
                    EthicalPrinciple.ECONOMIC_EFFICIENCY: 0.4,
                    EthicalPrinciple.CUSTOMER_SATISFACTION: 0.3,
                    EthicalPrinciple.WORKER_WELFARE: 0.2,
                    EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY: 0.1
                },
                "utility_function": "profit_sustainability_balance",
                "weight": 0.15
            },
            {
                "stakeholder_id": "REGULATOR",
                "stakeholder_type": "government",
                "priorities": {
                    EthicalPrinciple.SOCIAL_EQUITY: 0.3,
                    EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY: 0.3,
                    EthicalPrinciple.WORKER_WELFARE: 0.2,
                    EthicalPrinciple.ACCOUNTABILITY: 0.2
                },
                "utility_function": "compliance_equity_balance",
                "weight": 0.1
            }
        ]
        
        for config in stakeholders_config:
            stakeholder = Stakeholder(**config)
            self.stakeholders[stakeholder.stakeholder_id] = stakeholder
    
    def _initialize_ethical_constraints(self):
        """Initialize ethical constraints based on the constitution"""
        constraints = [
            EthicalConstraint(
                constraint_id="ENV_MAX_EMISSIONS",
                principle=EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY,
                constraint_type="upper_bound",
                threshold=100.0,
                penalty_weight=2.0,
                description="Maximum CO2 emissions per route"
            ),
            EthicalConstraint(
                constraint_id="WORK_MAX_HOURS",
                principle=EthicalPrinciple.WORKER_WELFARE,
                constraint_type="upper_bound",
                threshold=8.0,
                penalty_weight=1.5,
                description="Maximum working hours per driver"
            ),
            EthicalConstraint(
                constraint_id="CUST_MAX_WAIT",
                principle=EthicalPrinciple.CUSTOMER_SATISFACTION,
                constraint_type="upper_bound",
                threshold=4.0,
                penalty_weight=1.0,
                description="Maximum customer wait time"
            ),
            EthicalConstraint(
                constraint_id="ECON_MIN_MARGIN",
                principle=EthicalPrinciple.ECONOMIC_EFFICIENCY,
                constraint_type="lower_bound",
                threshold=0.15,
                penalty_weight=0.5,
                description="Minimum profit margin"
            ),
            EthicalConstraint(
                constraint_id="SOC_EQUITY_ACCESS",
                principle=EthicalPrinciple.SOCIAL_EQUITY,
                constraint_type="equality",
                threshold=0.2,
                penalty_weight=1.0,
                description="Equitable service access across areas"
            )
        ]
        
        self.ethical_constraints = constraints

print("✅ Constitutional AI system implemented")
```

```python
class ValueAlignedOptimizer:
    """Value-aligned optimizer balancing multiple ethical objectives"""
    
    def __init__(self, instance: VRPBInstance, constitutional_ai: ConstitutionalAI):
        self.instance = instance
        self.constitutional_ai = constitutional_ai
        self.optimization_history = []
        self.stakeholder_feedback = defaultdict(list)
        
        print(f"⚖️ Value-Aligned Optimizer initialized for {instance.num_customers} customers")
    
    def generate_ethical_alternatives(self) -> List[Dict[str, Any]]:
        """Generate routing alternatives with different ethical focus"""
        
        alternatives = []
        
        # Alternative 1: Economic Efficiency Focus (Route A from source)
        route_a = [1, 2, 3, 4, 5, 6, 7]
        metrics_a = {
            "total_distance": 25.0,
            "operating_cost": 150.0,
            "co2_emissions": 85.0,
            "driver_hours": 7.5,
            "driver_fatigue": 0.7,
            "max_wait_time": 3.5,
            "service_quality": 0.85,
            "profit_margin": 0.22,
            "service_variance": 0.25,
            "fuel_efficiency": 0.75,
            "decision_transparency": 0.8,
            "decision_traceability": 0.9
        }
        
        ethical_eval_a = self.constitutional_ai.evaluate_route_ethically(route_a, metrics_a)
        
        alternatives.append({
            "route_id": "A",
            "route": route_a,
            "description": "Economic efficiency focus - 15% cost savings",
            "metrics": metrics_a,
            "ethical_evaluation": ethical_eval_a,
            "focus": "economic_efficiency"
        })
        
        # Alternative 2: Environmental & Welfare Focus (Route B from source)
        route_b = [1, 3, 2, 4, 6, 5, 7]
        metrics_b = {
            "total_distance": 27.0,
            "operating_cost": 162.0,
            "co2_emissions": 51.0,
            "driver_hours": 6.8,
            "driver_fatigue": 0.4,
            "max_wait_time": 2.8,
            "service_quality": 0.92,
            "profit_margin": 0.18,
            "service_variance": 0.15,
            "fuel_efficiency": 0.88,
            "decision_transparency": 0.9,
            "decision_traceability": 0.95
        }
        
        ethical_eval_b = self.constitutional_ai.evaluate_route_ethically(route_b, metrics_b)
        
        alternatives.append({
            "route_id": "B",
            "route": route_b,
            "description": "Environmental & welfare focus - 40% lower emissions",
            "metrics": metrics_b,
            "ethical_evaluation": ethical_eval_b,
            "focus": "environmental_sustainability"
        })
        
        # Alternative 3: Balanced Approach
        route_c = [1, 4, 2, 3, 5, 7, 6]
        metrics_c = {
            "total_distance": 26.0,
            "operating_cost": 156.0,
            "co2_emissions": 68.0,
            "driver_hours": 7.2,
            "driver_fatigue": 0.55,
            "max_wait_time": 3.2,
            "service_quality": 0.88,
            "profit_margin": 0.20,
            "service_variance": 0.20,
            "fuel_efficiency": 0.82,
            "decision_transparency": 0.85,
            "decision_traceability": 0.92
        }
        
        ethical_eval_c = self.constitutional_ai.evaluate_route_ethically(route_c, metrics_c)
        
        alternatives.append({
            "route_id": "C",
            "route": route_c,
            "description": "Balanced approach - moderate cost and emissions",
            "metrics": metrics_c,
            "ethical_evaluation": ethical_eval_c,
            "focus": "balanced"
        })
        
        return alternatives

print("✅ Value-Aligned Optimizer implemented")
```

```python
# Initialize the Value-Aligned & Ethical Framework
constitutional_ai = ConstitutionalAI(instance)
optimizer = ValueAlignedOptimizer(instance, constitutional_ai)

print("\n" + "="*60)
print("🏛️ VALUE-ALIGNED & ETHICAL FRAMEWORK INITIALIZATION")
print("="*60)

print(f"\n📋 Ethical Constitution:")
for principle, config in constitutional_ai.constitution.items():
    print(f"   {principle.value.replace('_', ' ').title()}:")
    print(f"      Weight: {config['weight']:.2f}")
    print(f"      Description: {config['description']}")

print(f"\n👥 Stakeholders:")
for stakeholder_id, stakeholder in constitutional_ai.stakeholders.items():
    print(f"   {stakeholder_id} ({stakeholder.stakeholder_type}):")
    print(f"      Weight: {stakeholder.weight:.2f}")
    print(f"      Top Priority: {max(stakeholder.priorities, key=stakeholder.priorities.get).value.replace('_', ' ')}")

print(f"\n⚖️ Ethical Constraints:")
for constraint in constitutional_ai.ethical_constraints:
    print(f"   {constraint.constraint_id}: {constraint.description}")
    print(f"      Threshold: {constraint.threshold}, Type: {constraint.constraint_type}")
    print(f"      Penalty Weight: {constraint.penalty_weight}")
```

```python
try:
    # Generate and evaluate ethical routing alternatives
    print("\n" + "="*60)
    print("🔄 ETHICAL ROUTING OPTIMIZATION")
    print("="*60)
    
    # Generate alternatives
    alternatives = optimizer.generate_ethical_alternatives()
    
    print(f"\n📊 Generated {len(alternatives)} routing alternatives:")
    for alt in alternatives:
        print(f"\n--- Route {alt['route_id']}: {alt['description']} ---")
        print(f"   Route: {alt['route']}")
        print(f"   Cost: ${alt['metrics']['operating_cost']:.0f}, Emissions: {alt['metrics']['co2_emissions']:.1f} kg CO2")
        print(f"   Driver Hours: {alt['metrics']['driver_hours']:.1f}, Customer Wait: {alt['metrics']['max_wait_time']:.1f} hrs")
        print(f"   Overall Ethical Score: {alt['ethical_evaluation']['overall_ethical_score']:.3f}")
        
        # Stakeholder utilities
        print(f"   Stakeholder Utilities:")
        for stakeholder_id, utility in alt['ethical_evaluation']['stakeholder_utilities'].items():
            stakeholder = constitutional_ai.stakeholders[stakeholder_id]
            print(f"      {stakeholder_id}: {utility:.3f}")
        
        # Constraint violations
        if alt['ethical_evaluation']['violations']:
            print(f"   Constraint Violations:")
            for violation in alt['ethical_evaluation']['violations']:
                print(f"      {violation['constraint_id']}: Severity {violation['severity']:.2f}")
        else:
            print(f"   ✅ All constraints satisfied")
        
        # Bias detections
        if alt['ethical_evaluation']['bias_detections']:
            print(f"   Bias Detections:")
            for bias in alt['ethical_evaluation']['bias_detections']:
                print(f"      {bias.bias_type}: Score {bias.bias_score:.2f}")
        else:
            print(f"   ✅ No biases detected")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    def visualize_ethical_framework(alternatives: List[Dict[str, Any]], 
                                    selected: Dict[str, Any],
                                    constitutional_ai: ConstitutionalAI):
        """Create comprehensive visualization of ethical framework performance"""
        
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
        fig.suptitle('Value-Aligned & Ethical Framework Performance', fontsize=16, fontweight='bold')
        
        # 1. Ethical Score Comparison
        route_ids = [alt['route_id'] for alt in alternatives]
        ethical_scores = [alt['ethical_evaluation']['overall_ethical_score'] for alt in alternatives]
        
        colors = ['lightcoral' if alt['route_id'] == selected['route_id'] else 'skyblue' for alt in alternatives]
        bars = ax1.bar(route_ids, ethical_scores, color=colors)
        ax1.set_ylabel('Ethical Score')
        ax1.set_title('Overall Ethical Score by Route')
        ax1.set_ylim(0, 1)
        
        # Add value labels on bars
        for bar, score in zip(bars, ethical_scores):
            height = bar.get_height()
            ax1.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                    f'{score:.3f}', ha='center', va='bottom', fontsize=10)
        
        # 2. Stakeholder Utility Heatmap
        stakeholder_ids = list(constitutional_ai.stakeholders.keys())
        utility_matrix = []
        
        for alt in alternatives:
            utilities = alt['ethical_evaluation']['stakeholder_utilities']
            utility_row = [utilities.get(s_id, 0.5) for s_id in stakeholder_ids]
            utility_matrix.append(utility_row)
        
        utility_matrix = np.array(utility_matrix).T
        
        im = ax2.imshow(utility_matrix, cmap='RdYlGn', aspect='auto', vmin=0, vmax=1)
        ax2.set_xticks(range(len(route_ids)))
        ax2.set_xticklabels(route_ids)
        ax2.set_yticks(range(len(stakeholder_ids)))
        ax2.set_yticklabels(stakeholder_ids)
        ax2.set_title('Stakeholder Utility Heatmap')
        
        # Add colorbar
        cbar = plt.colorbar(im, ax=ax2)
        cbar.set_label('Utility Score')
        
        # 3. Ethical Principle Performance
        principle_names = []
        principle_scores_selected = []
        principle_scores_economic = []
        
        selected_eval = selected['ethical_evaluation']
        economic_alt = next(alt for alt in alternatives if alt['focus'] == 'economic_efficiency')
        economic_eval = economic_alt['ethical_evaluation']
        
        for principle_id, score in selected_eval['principle_scores'].items():
            principle_names.append(principle_id.replace('_', '\n').title())
            principle_scores_selected.append(score)
            principle_scores_economic.append(economic_eval['principle_scores'][principle_id])
        
        x = np.arange(len(principle_names))
        width = 0.35
        
        ax3.bar(x - width/2, principle_scores_selected, width, 
                label='Ethical Solution', color='lightgreen')
        ax3.bar(x + width/2, principle_scores_economic, width, 
                label='Economic Solution', color='lightcoral')
        
        ax3.set_xlabel('Ethical Principles')
        ax3.set_ylabel('Principle Score')
        ax3.set_title('Ethical vs Economic Solution Comparison')
        ax3.set_xticks(x)
        ax3.set_xticklabels(principle_names, rotation=45, ha='right')
        ax3.legend()
        ax3.set_ylim(0, 1)
        ax3.grid(True, alpha=0.3)
        
        # 4. Cost vs Emissions Trade-off
        costs = [alt['metrics']['operating_cost'] for alt in alternatives]
        emissions = [alt['metrics']['co2_emissions'] for alt in alternatives]
        
        for i, alt in enumerate(alternatives):
            color = 'red' if alt['route_id'] == selected['route_id'] else 'blue'
            ax4.scatter(costs[i], emissions[i], s=100, c=color, label=alt['route_id'])
            ax4.annotate(alt['route_id'], (costs[i], emissions[i]), 
                        xytext=(5, 5), textcoords='offset points')
        
        ax4.set_xlabel('Operating Cost ($)')
        ax4.set_ylabel('CO2 Emissions (kg)')
        ax4.set_title('Cost vs Environmental Impact Trade-off')
        ax4.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
    
    visualize_ethical_framework(alternatives, selected_alternative, constitutional_ai)
except Exception as e:
    print(f'Error in cell: {e}')
```
