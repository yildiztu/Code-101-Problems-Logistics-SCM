## Tier 8: Value-Aligned & Ethical Framework (Constitutional AI)

### Key assumptions
- Freight mode decisions must align with corporate values and ethical principles
- Multiple stakeholders have legitimate interests in transportation decisions
- Environmental sustainability, social responsibility, and economic viability are balanced
- Ethical constraints are encoded as hard requirements in optimization models
- Decision transparency and accountability are mandatory for stakeholder trust

### Approach (step-by-step)
1. **Define constitutional principles**: Establish core values and ethical constraints
2. **Identify stakeholders**: Map all affected parties and their interests
3. **Create ethical scoring system**: Quantify social and environmental impacts
4. **Implement multi-objective optimization**: Balance cost, service, sustainability, and ethics
5. **Add compliance verification**: Ensure all decisions meet ethical requirements
6. **Enable stakeholder feedback**: Incorporate stakeholder preferences into decisions
7. **Provide transparency reports**: Document ethical considerations in all decisions

### What to look for in the results
- Ethical constraint satisfaction in all freight mode decisions
- Multi-stakeholder utility optimization across different interest groups
- Environmental impact minimization while maintaining service quality
- Social responsibility considerations in transportation choices
- Transparent decision documentation with ethical reasoning
- Trade-off analysis between economic efficiency and ethical objectives

```python
class EthicalScoringSystem:
    """
    Ethical scoring system for transportation mode evaluation
    Quantifies social and environmental impacts
    """
    
    def __init__(self, params: TransportParams, stakeholders: List[Stakeholder]):
        self.params = params
        self.stakeholders = stakeholders
        self.stakeholder_map = {s.stakeholder_id: s for s in stakeholders}
        
        # Ethical scoring parameters
        self.carbon_impact_weight = 0.3
        self.energy_impact_weight = 0.2
        self.social_impact_weight = 0.3
        self.economic_impact_weight = 0.2
        
        # Impact baselines for normalization
        self.carbon_baseline = 500.0  # kg CO2
        self.energy_baseline = 1.0     # kWh per kg
        self.social_baseline = 0.5    # Normalized social impact
        self.economic_baseline = 1000.0 # USD cost
    
    def calculate_environmental_impact(self, shipment: Shipment, mode: str) -> Dict[str, float]:
        """
        Calculate environmental impact for shipment and mode
        """
        destination = shipment.destination
        
        # Carbon emissions
        carbon_emissions = shipment.weight * self.params.carbon_emissions[mode][destination]
        carbon_score = self._normalize_impact(carbon_emissions, self.carbon_baseline, True)  # Lower is better
        
        # Energy consumption
        energy_consumption = shipment.weight * self.params.energy_consumption[mode][destination]
        energy_score = self._normalize_impact(energy_consumption, self.energy_baseline * shipment.weight, True)
        
        # Additional environmental factors
        if mode == "ocean":
            # Ocean freight has additional concerns
            ballast_water_impact = 0.1 if shipment.weight > 2000 else 0.05
            marine_life_risk = 0.08 if "hazardous" in shipment.special_requirements else 0.02
        else:  # air
            # Air freight has different concerns
            air_pollution_impact = 0.15
            contrail_effect = 0.05
            ballast_water_impact = 0.0
            marine_life_risk = 0.0
        
        return {
            "carbon_emissions": carbon_emissions,
            "carbon_score": carbon_score,
            "energy_consumption": energy_consumption,
            "energy_score": energy_score,
            "air_pollution": air_pollution_impact if mode == "air" else 0.02,
            "ballast_water": ballast_water_impact,
            "marine_life_risk": marine_life_risk,
            "overall_environmental_score": (carbon_score + energy_score) / 2
        }
    
    def calculate_social_impact(self, shipment: Shipment, mode: str) -> Dict[str, float]:
        """
        Calculate social impact for shipment and mode
        """
        # Labor standards impact
        if mode == "ocean":
            labor_standards_score = 0.8  # Generally good standards in ocean shipping
            job_creation_impact = 0.6   # Moderate job creation
            working_conditions = 0.7    # Decent working conditions
        else:  # air
            labor_standards_score = 0.9  # Higher standards in air freight
            job_creation_impact = 0.4   # Fewer jobs per ton
            working_conditions = 0.85   # Better working conditions
        
        # Community impact
        destination = shipment.destination
        if destination in ["hamburg", "rotterdam"]:
            community_impact = 0.4 if mode == "ocean" else 0.2  # Ocean freight supports port communities
            noise_pollution = 0.3 if mode == "ocean" else 0.7  # Air freight noisier
        else:
            community_impact = 0.3 if mode == "ocean" else 0.1
            noise_pollution = 0.2 if mode == "ocean" else 0.5
        
        # Special requirements social impact
        special_requirements_impact = 0.0
        if "hazardous" in shipment.special_requirements:
            special_requirements_impact = 0.3 if mode == "ocean" else 0.4  # Higher risk for air
        if "temperature_controlled" in shipment.special_requirements:
            special_requirements_impact += 0.1  # Additional handling requirements
        
        return {
            "labor_standards": labor_standards_score,
            "job_creation": job_creation_impact,
            "working_conditions": working_conditions,
            "community_impact": community_impact,
            "noise_pollution": noise_pollution,
            "special_requirements": special_requirements_impact,
            "overall_social_score": (labor_standards_score + job_creation_impact + 
                                    working_conditions + community_impact - noise_pollution) / 4
        }
    
    def calculate_stakeholder_utilities(self, shipment: Shipment, mode: str, 
                                        environmental_impact: Dict, social_impact: Dict, 
                                        economic_cost: float) -> Dict[str, float]:
        """
        Calculate utility for each stakeholder
        """
        utilities = {}
        
        for stakeholder in self.stakeholders:
            utility = 0.0
            
            # Calculate utility based on stakeholder interests
            for interest, weight in stakeholder.utility_weights.items():
                if interest == "cost" or interest == "profit":
                    # Lower cost is better for customers, higher profit for shareholders
                    if stakeholder.stakeholder_type == "shareholder":
                        # Profit is inverse of cost (simplified)
                        cost_utility = self._normalize_impact(economic_cost, self.economic_baseline, True)
                        utility += weight * cost_utility
                    else:
                        cost_utility = self._normalize_impact(economic_cost, self.economic_baseline, True)
                        utility += weight * cost_utility
                
                elif interest == "service" or interest == "reliability":
                    # Service quality based on transit time
                    transit_time = self.params.transit_times[mode][shipment.destination]
                    service_utility = self._normalize_impact(transit_time, 15, True)  # Lower time is better
                    utility += weight * service_utility
                
                elif interest == "environment" or interest == "carbon_emissions":
                    # Environmental utility
                    env_utility = environmental_impact["overall_environmental_score"]
                    utility += weight * env_utility
                
                elif interest == "labor_standards" or interest == "safety":
                    # Labor and safety utility
                    labor_utility = social_impact["labor_standards"]
                    utility += weight * labor_utility
                
                elif interest == "local_jobs" or interest == "community":
                    # Community impact utility
                    community_utility = social_impact["community_impact"]
                    utility += weight * community_utility
                
                elif interest == "noise":
                    # Noise pollution (lower is better)
                    noise_utility = 1.0 - social_impact["noise_pollution"]
                    utility += weight * noise_utility
                
                elif interest == "energy_efficiency":
                    # Energy efficiency utility
                    energy_utility = environmental_impact["energy_score"]
                    utility += weight * energy_utility
            
            # Apply stakeholder importance weighting from shipment
            stakeholder_importance = shipment.stakeholder_importance.get(stakeholder.stakeholder_type, 0.1)
            utilities[stakeholder.stakeholder_id] = utility * stakeholder_importance
        
        return utilities
    
    def calculate_overall_ethical_score(self, environmental_impact: Dict, 
                                        social_impact: Dict, 
                                        stakeholder_utilities: Dict) -> float:
        """
        Calculate overall ethical performance score
        """
        # Environmental component
        env_score = environmental_impact["overall_environmental_score"]
        
        # Social component
        social_score = social_impact["overall_social_score"]
        
        # Stakeholder utility component
        avg_utility = np.mean(list(stakeholder_utilities.values()))
        
        # Weighted combination
        overall_score = (
            self.carbon_impact_weight * env_score +
            self.energy_impact_weight * env_score +  # Energy included in env_score
            self.social_impact_weight * social_score +
            self.economic_impact_weight * avg_utility
        )
        
        return max(0.0, min(1.0, overall_score))
    
    def _normalize_impact(self, value: float, baseline: float, lower_is_better: bool) -> float:
        """
        Normalize impact to 0-1 scale
        """
        if lower_is_better:
            # Lower values are better (e.g., cost, emissions)
            return max(0.0, min(1.0, baseline / max(value, 0.1)))
        else:
            # Higher values are better (e.g., utility)
            return max(0.0, min(1.0, value / baseline))

# Initialize ethical scoring system
ethical_scorer = EthicalScoringSystem(params, stakeholders)
print("Ethical Scoring System initialized:")
print(f"  Environmental Weight: {ethical_scorer.carbon_impact_weight + ethical_scorer.energy_impact_weight:.1%}")
print(f"  Social Weight: {ethical_scorer.social_impact_weight:.1%}")
print(f"  Economic Weight: {ethical_scorer.economic_impact_weight:.1%}")
```

```python
class EthicalConstraintChecker:
    """
    Ethical constraint verification and compliance checking
    """
    
    def __init__(self, constraints: List[EthicalConstraint]):
        self.constraints = constraints
        self.constraint_map = {c.constraint_id: c for c in constraints}
        self.compliance_history = []
        
    def check_compliance(self, shipment: Shipment, mode: str, 
                        environmental_impact: Dict, social_impact: Dict) -> Dict[str, bool]:
        """
        Check compliance with all ethical constraints
        """
        compliance_results = {}
        
        for constraint in self.constraints:
            compliant = self._check_single_constraint(
                constraint, shipment, mode, environmental_impact, social_impact
            )
            compliance_results[constraint.constraint_id] = compliant
        
        # Record compliance check
        self.compliance_history.append({
            "shipment_id": shipment.id,
            "mode": mode,
            "compliance": compliance_results,
            "timestamp": datetime.now()
        })
        
        return compliance_results
    
    def _check_single_constraint(self, constraint: EthicalConstraint, shipment: Shipment, 
                               mode: str, environmental_impact: Dict, social_impact: Dict) -> bool:
        """
        Check compliance with a single constraint
        """
        # Check if constraint applies to this destination
        if (constraint.applicable_destinations and 
            shipment.destination not in constraint.applicable_destinations):
            return True  # Constraint doesn't apply, automatically compliant
        
        # Check based on constraint metric
        if constraint.metric == "carbon_emissions":
            carbon_emissions = environmental_impact["carbon_emissions"]
            return carbon_emissions <= constraint.threshold
        
        elif constraint.metric == "labor_standards":
            labor_standards = social_impact["labor_standards"]
            return labor_standards >= constraint.threshold
        
        elif constraint.metric == "noise_pollution":
            noise_pollution = social_impact["noise_pollution"] * 100  # Convert to 0-100 scale
            return noise_pollution <= constraint.threshold
        
        elif constraint.metric == "energy_efficiency":
            energy_efficiency = environmental_impact["energy_consumption"] / shipment.weight
            return energy_efficiency <= constraint.threshold
        
        elif constraint.metric == "community_impact":
            community_impact = social_impact["community_impact"]
            return community_impact <= constraint.threshold
        
        else:
            return True  # Unknown metric, assume compliant
    
    def calculate_compliance_score(self, compliance_results: Dict[str, bool]) -> float:
        """
        Calculate overall compliance score
        """
        if not compliance_results:
            return 1.0
        
        compliant_count = sum(1 for compliant in compliance_results.values() if compliant)
        total_count = len(compliance_results)
        
        return compliant_count / total_count
    
    def get_violation_penalty(self, compliance_results: Dict[str, bool]) -> float:
        """
        Calculate penalty for constraint violations
        """
        penalty = 0.0
        
        for constraint_id, compliant in compliance_results.items():
            if not compliant:
                constraint = self.constraint_map[constraint_id]
                if constraint.constraint_type == "hard_constraint":
                    penalty += constraint.penalty_weight  # Large penalty for hard constraints
                elif constraint.constraint_type == "soft_constraint":
                    penalty += constraint.penalty_weight * 0.5  # Moderate penalty for soft constraints
                # Guidance principles have no penalty, just preference
        
        return penalty
    
    def get_compliance_summary(self) -> Dict:
        """
        Get summary of compliance history
        """
        if not self.compliance_history:
            return {"total_checks": 0, "compliance_rate": 0.0}
        
        total_checks = len(self.compliance_history) * len(self.constraints)
        compliant_checks = 0
        
        for check in self.compliance_history:
            compliant_checks += sum(1 for compliant in check["compliance"].values() if compliant)
        
        compliance_rate = compliant_checks / total_checks if total_checks > 0 else 0.0
        
        return {
            "total_checks": total_checks,
            "compliant_checks": compliant_checks,
            "compliance_rate": compliance_rate,
            "constraint_compliance": self._calculate_constraint_specific_compliance()
        }
    
    def _calculate_constraint_specific_compliance(self) -> Dict[str, float]:
        """
        Calculate compliance rate for each specific constraint
        """
        constraint_compliance = {}
        
        for constraint in self.constraints:
            compliant_count = 0
            total_count = 0
            
            for check in self.compliance_history:
                if constraint.constraint_id in check["compliance"]:
                    total_count += 1
                    if check["compliance"][constraint.constraint_id]:
                        compliant_count += 1
            
            compliance_rate = compliant_count / total_count if total_count > 0 else 0.0
            constraint_compliance[constraint.constraint_id] = compliance_rate
        
        return constraint_compliance

# Initialize ethical constraint checker
constraint_checker = EthicalConstraintChecker(ethical_constraints)
print("Ethical Constraint Checker initialized:")
print(f"  Total Constraints: {len(constraint_checker.constraints)}")
print(f"  Hard Constraints: {len([c for c in ethical_constraints if c.constraint_type == 'hard_constraint'])}")
print(f"  Soft Constraints: {len([c for c in ethical_constraints if c.constraint_type == 'soft_constraint'])}")
print(f"  Guidance Principles: {len([c for c in ethical_constraints if c.constraint_type == 'guidance_principle'])}")
```

```python
class ValueAlignedOptimizer:
    """
    Value-aligned freight mode optimizer with ethical constraints
    Multi-objective optimization balancing economic, environmental, and social objectives
    """
    
    def __init__(self, params: TransportParams, stakeholders: List[Stakeholder], 
                 ethical_constraints: List[EthicalConstraint]):
        self.params = params
        self.stakeholders = stakeholders
        self.ethical_constraints = ethical_constraints
        
        # Initialize components
        self.ethical_scorer = EthicalScoringSystem(params, stakeholders)
        self.constraint_checker = EthicalConstraintChecker(ethical_constraints)
        
        # Optimization parameters
        self.economic_weight = 0.3
        self.environmental_weight = 0.35
        self.social_weight = 0.35
        
        # Decision history
        self.decision_history = []
        self.performance_metrics = {
            "economic_performance": [],
            "environmental_performance": [],
            "social_performance": [],
            "ethical_compliance": [],
            "stakeholder_satisfaction": []
        }
    
    def optimize_freight_mode(self, shipment: Shipment) -> EthicalDecision:
        """
        Optimize freight mode with ethical considerations
        """
        # Evaluate both modes
        ocean_evaluation = self._evaluate_mode(shipment, "ocean")
        air_evaluation = self._evaluate_mode(shipment, "air")
        
        # Check for hard constraint violations
        ocean_violates_hard = self._violates_hard_constraints(ocean_evaluation["compliance"])
        air_violates_hard = self._violates_hard_constraints(air_evaluation["compliance"])
        
        # Eliminate options that violate hard constraints
        viable_options = []
        if not ocean_violates_hard:
            viable_options.append(("ocean", ocean_evaluation))
        if not air_violates_hard:
            viable_options.append(("air", air_evaluation))
        
        # If no viable options, choose least violation
        if not viable_options:
            # Choose option with minimum hard constraint violations
            ocean_hard_violations = self._count_hard_violations(ocean_evaluation["compliance"])
            air_hard_violations = self._count_hard_violations(air_evaluation["compliance"])
            
            if ocean_hard_violations <= air_hard_violations:
                chosen_mode = "ocean"
                chosen_evaluation = ocean_evaluation
            else:
                chosen_mode = "air"
                chosen_evaluation = air_evaluation
        else:
            # Choose best option among viable ones
            best_mode, best_evaluation = self._select_best_option(viable_options)
            chosen_mode = best_mode
            chosen_evaluation = best_evaluation
        
        # Create ethical decision
        ethical_decision = self._create_ethical_decision(
            shipment, chosen_mode, chosen_evaluation
        )
        
        # Record decision
        self.decision_history.append(ethical_decision)
        self._update_performance_metrics(ethical_decision)
        
        return ethical_decision
    
    def _evaluate_mode(self, shipment: Shipment, mode: str) -> Dict:
        """
        Evaluate transportation mode comprehensively
        """
        # Calculate economic cost
        economic_cost = self._calculate_total_cost(shipment, mode)
        
        # Calculate environmental impact
        environmental_impact = self.ethical_scorer.calculate_environmental_impact(shipment, mode)
        
        # Calculate social impact
        social_impact = self.ethical_scorer.calculate_social_impact(shipment, mode)
        
        # Calculate stakeholder utilities
        stakeholder_utilities = self.ethical_scorer.calculate_stakeholder_utilities(
            shipment, mode, environmental_impact, social_impact, economic_cost
        )
        
        # Check ethical compliance
        compliance = self.constraint_checker.check_compliance(
            shipment, mode, environmental_impact, social_impact
        )
        
        # Calculate overall score
        ethical_score = self.ethical_scorer.calculate_overall_ethical_score(
            environmental_impact, social_impact, stakeholder_utilities
        )
        
        # Calculate compliance penalty
        compliance_penalty = self.constraint_checker.get_violation_penalty(compliance)
        
        # Final objective score
        objective_score = (
            self.economic_weight * self._normalize_cost(economic_cost) +
            self.environmental_weight * environmental_impact["overall_environmental_score"] +
            self.social_weight * social_impact["overall_social_score"] -
            compliance_penalty * 0.1  # Penalty reduces score
        )
        
        return {
            "mode": mode,
            "economic_cost": economic_cost,
            "environmental_impact": environmental_impact,
            "social_impact": social_impact,
            "stakeholder_utilities": stakeholder_utilities,
            "compliance": compliance,
            "ethical_score": ethical_score,
            "compliance_penalty": compliance_penalty,
            "objective_score": objective_score
        }
    
    def _calculate_total_cost(self, shipment: Shipment, mode: str) -> float:
        """
        Calculate total economic cost
        """
        # Transportation cost
        transport_cost = shipment.weight * self.params.costs[mode][shipment.destination]
        
        # Transit time
        transit_time = self.params.transit_times[mode][shipment.destination]
        
        # Holding cost
        if transit_time < shipment.due_date:
            holding_cost = (shipment.due_date - transit_time) * self.params.holding_rate * shipment.value
        else:
            holding_cost = 0
        
        # Late penalty
        if transit_time > shipment.due_date:
            late_days = transit_time - shipment.due_date
            late_penalty_cost = late_days * self.params.late_penalty
        else:
            late_penalty_cost = 0
        
        # Special requirements cost
        special_cost = 0
        if "temperature_controlled" in shipment.special_requirements:
            special_cost += shipment.weight * 0.5
        if "hazardous" in shipment.special_requirements:
            special_cost += shipment.weight * 1.2
        
        return transport_cost + holding_cost + late_penalty_cost + special_cost
    
    def _normalize_cost(self, cost: float) -> float:
        """
        Normalize cost to 0-1 scale (lower cost = higher score)
        """
        baseline_cost = 5000.0  # Reasonable baseline
        return max(0.0, min(1.0, baseline_cost / max(cost, 100.0)))
    
    def _violates_hard_constraints(self, compliance: Dict[str, bool]) -> bool:
        """
        Check if any hard constraints are violated
        """
        for constraint_id, compliant in compliance.items():
            constraint = self.constraint_checker.constraint_map[constraint_id]
            if constraint.constraint_type == "hard_constraint" and not compliant:
                return True
        return False
    
    def _count_hard_violations(self, compliance: Dict[str, bool]) -> int:
        """
        Count hard constraint violations
        """
        violations = 0
        for constraint_id, compliant in compliance.items():
            constraint = self.constraint_checker.constraint_map[constraint_id]
            if constraint.constraint_type == "hard_constraint" and not compliant:
                violations += 1
        return violations
    
    def _select_best_option(self, viable_options: List[Tuple[str, Dict]]) -> Tuple[str, Dict]:
        """
        Select best option among viable alternatives
        """
        best_option = None
        best_score = -1.0
        
        for mode, evaluation in viable_options:
            if evaluation["objective_score"] > best_score:
                best_score = evaluation["objective_score"]
                best_option = (mode, evaluation)
        
        return best_option
    
    def _create_ethical_decision(self, shipment: Shipment, mode: str, evaluation: Dict) -> EthicalDecision:
        """
        Create ethical decision object
        """
        # Generate transparency report
        transparency_report = self._generate_transparency_report(shipment, mode, evaluation)
        
        # Generate alternative analysis
        alternatives = self._generate_alternatives_analysis(shipment, mode)
        
        return EthicalDecision(
            shipment_id=shipment.id,
            chosen_mode=mode,
            economic_cost=evaluation["economic_cost"],
            environmental_impact=evaluation["environmental_impact"],
            social_impact=evaluation["social_impact"],
            stakeholder_utilities=evaluation["stakeholder_utilities"],
            ethical_compliance=evaluation["compliance"],
            ethical_score=evaluation["ethical_score"],
            transparency_report=transparency_report,
            alternative_considered=alternatives
        )
    
    def _generate_transparency_report(self, shipment: Shipment, mode: str, evaluation: Dict) -> str:
        """
        Generate detailed transparency report for decision
        """
        report_parts = []
        
        report_parts.append(f"Decision for Shipment {shipment.id} to {shipment.destination}:")
        report_parts.append(f"Chosen Mode: {mode.upper()}")
        report_parts.append(f"Economic Cost: ${evaluation['economic_cost']:,.0f}")
        report_parts.append(f"Environmental Score: {evaluation['environmental_impact']['overall_environmental_score']:.1%}")
        report_parts.append(f"Social Score: {evaluation['social_impact']['overall_social_score']:.1%}")
        report_parts.append(f"Ethical Compliance: {self.constraint_checker.calculate_compliance_score(evaluation['compliance']):.1%}")
        
        # Add key factors
        report_parts.append("\nKey Decision Factors:")
        
        if evaluation["environmental_impact"]["carbon_emissions"] > 800:
            report_parts.append("- High carbon emissions were a significant concern")
        
        if shipment.priority == "high" and mode == "air":
            report_parts.append("- High priority justified air transport for service reliability")
        
        if "hazardous" in shipment.special_requirements and mode == "ocean":
            report_parts.append("- Ocean transport chosen for safer hazardous material handling")
        
        # Add compliance status
        report_parts.append("\nEthical Compliance Status:")
        for constraint_id, compliant in evaluation["compliance"].items():
            constraint = self.constraint_checker.constraint_map[constraint_id]
            status = "COMPLIANT" if compliant else "VIOLATION"
            report_parts.append(f"- {constraint.name}: {status}")
        
        # Add stakeholder impact
        report_parts.append("\nStakeholder Impact Summary:")
        for stakeholder_id, utility in evaluation["stakeholder_utilities"].items():
            stakeholder = self.stakeholder_map[stakeholder_id]
            report_parts.append(f"- {stakeholder.name}: Utility {utility:.2f}")
        
        return "\n".join(report_parts)
    
    def _generate_alternatives_analysis(self, shipment: Shipment, chosen_mode: str) -> List[Dict]:
        """
        Generate analysis of alternative modes
        """
        alternatives = []
        
        for mode in ["ocean", "air"]:
            if mode != chosen_mode:
                evaluation = self._evaluate_mode(shipment, mode)
                alternatives.append({
                    "mode": mode,
                    "economic_cost": evaluation["economic_cost"],
                    "environmental_score": evaluation["environmental_impact"]["overall_environmental_score"],
                    "social_score": evaluation["social_impact"]["overall_social_score"],
                    "ethical_compliance": self.constraint_checker.calculate_compliance_score(evaluation["compliance"]),
                    "reason_for_rejection": self._get_rejection_reason(chosen_mode, mode, evaluation)
                })
        
        return alternatives
    
    def _get_rejection_reason(self, chosen_mode: str, alternative_mode: str, evaluation: Dict) -> str:
        """
        Get reason for rejecting alternative mode
        """
        reasons = []
        
        # Check for hard constraint violations
        if self._violates_hard_constraints(evaluation["compliance"]):
            reasons.append("Violates hard ethical constraints")
        
        # Check cost difference
        # (This would need the chosen mode evaluation for accurate comparison)
        reasons.append("Lower overall ethical score")
        
        return ", ".join(reasons) if reasons else "Lower overall performance score"
    
    def _update_performance_metrics(self, decision: EthicalDecision):
        """
        Update performance metrics
        """
        # Economic performance (normalized)
        economic_perf = self._normalize_cost(decision.economic_cost)
        self.performance_metrics["economic_performance"].append(economic_perf)
        
        # Environmental performance
        env_perf = decision.environmental_impact["overall_environmental_score"]
        self.performance_metrics["environmental_performance"].append(env_perf)
        
        # Social performance
        social_perf = decision.social_impact["overall_social_score"]
        self.performance_metrics["social_performance"].append(social_perf)
        
        # Ethical compliance
        compliance_rate = self.constraint_checker.calculate_compliance_score(decision.ethical_compliance)
        self.performance_metrics["ethical_compliance"].append(compliance_rate)
        
        # Stakeholder satisfaction
        avg_utility = np.mean(list(decision.stakeholder_utilities.values()))
        self.performance_metrics["stakeholder_satisfaction"].append(avg_utility)
    
    def get_performance_summary(self) -> Dict:
        """
        Get performance summary
        """
        if not self.decision_history:
            return {"error": "No decisions made yet"}
        
        return {
            "total_decisions": len(self.decision_history),
            "average_economic_performance": np.mean(self.performance_metrics["economic_performance"]),
            "average_environmental_performance": np.mean(self.performance_metrics["environmental_performance"]),
            "average_social_performance": np.mean(self.performance_metrics["social_performance"]),
            "average_ethical_compliance": np.mean(self.performance_metrics["ethical_compliance"]),
            "average_stakeholder_satisfaction": np.mean(self.performance_metrics["stakeholder_satisfaction"]),
            "compliance_summary": self.constraint_checker.get_compliance_summary()
        }

# Initialize value-aligned optimizer
optimizer = ValueAlignedOptimizer(params, stakeholders, ethical_constraints)
print("\nValue-Aligned Optimizer initialized:")
print(f"  Economic Weight: {optimizer.economic_weight:.1%}")
print(f"  Environmental Weight: {optimizer.environmental_weight:.1%}")
print(f"  Social Weight: {optimizer.social_weight:.1%}")
print(f"  Ethical Constraints: {len(optimizer.ethical_constraints)}")
```

```python
# Demonstrate stakeholder analysis and trade-off visualization
def demonstrate_stakeholder_analysis(ethical_decisions: List[EthicalDecision], 
                                   stakeholders: List[Stakeholder]):
    """
    Demonstrate detailed stakeholder analysis and trade-offs
    """
    print("\nDemonstrating Stakeholder Analysis and Trade-Offs...")
    print("=" * 80)
    
    # Calculate stakeholder satisfaction by type
    stakeholder_satisfaction = {s.stakeholder_type: [] for s in stakeholders}
    
    for decision in ethical_decisions:
        for stakeholder in stakeholders:
            utility = decision.stakeholder_utilities[stakeholder.stakeholder_id]
            stakeholder_satisfaction[stakeholder.stakeholder_type].append(utility)
    
    print("\nStakeholder Satisfaction by Type:")
    for stakeholder_type, utilities in stakeholder_satisfaction.items():
        avg_utility = np.mean(utilities)
        std_utility = np.std(utilities)
        print(f"  {stakeholder_type.title()}: {avg_utility:.3f} ± {std_utility:.3f}")
    
    # Analyze trade-offs
    print("\nTrade-Off Analysis:")
    
    # Economic vs Environmental trade-off
    economic_scores = [optimizer._normalize_cost(d.economic_cost) for d in ethical_decisions]
    environmental_scores = [d.environmental_impact['overall_environmental_score'] for d in ethical_decisions]
    
    correlation = np.corrcoef(economic_scores, environmental_scores)[0, 1]
    print(f"  Economic vs Environmental Correlation: {correlation:.3f}")
    
    if correlation < -0.3:
        print(f"    → Strong trade-off between cost and environmental performance")
    elif correlation < 0.3:
        print(f"    → Weak correlation between cost and environmental performance")
    else:
        print(f"    → Cost and environmental performance aligned")
    
    # Social vs Environmental trade-off
    social_scores = [d.social_impact['overall_social_score'] for d in ethical_decisions]
    social_env_correlation = np.corrcoef(social_scores, environmental_scores)[0, 1]
    print(f"  Social vs Environmental Correlation: {social_env_correlation:.3f}")
    
    # Identify decisions with significant trade-offs
    print("\nDecisions with Significant Trade-Offs:")
    trade_off_threshold = 0.3  # 30% difference between objectives
    
    for decision in ethical_decisions:
        econ_score = optimizer._normalize_cost(decision.economic_cost)
        env_score = decision.environmental_impact['overall_environmental_score']
        soc_score = decision.social_impact['overall_social_score']
        
        max_score = max(econ_score, env_score, soc_score)
        min_score = min(econ_score, env_score, soc_score)
        
        if max_score - min_score > trade_off_threshold:
            print(f"\n  Shipment {decision.shipment.id} ({decision.chosen_mode.upper()}):")
            print(f"    Economic: {econ_score:.1%}, Environmental: {env_score:.1%}, Social: {soc_score:.1%}")
            print(f"    Range: {max_score - min_score:.1%} (significant trade-off)")
            
            # Identify which objective was sacrificed
            if econ_score == min_score:
                print(f"    → Economic performance sacrificed for environmental/social goals")
            elif env_score == min_score:
                print(f"    → Environmental performance sacrificed for economic/social goals")
            else:
                print(f"    → Social performance sacrificed for economic/environmental goals")

# Run stakeholder analysis
demonstrate_stakeholder_analysis(ethical_decisions, stakeholders)

# Create stakeholder trade-off visualization
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Stakeholder Analysis and Ethical Trade-Offs', fontsize=16, fontweight='bold')

# 1. Stakeholder utility heatmap
stakeholder_matrix = []
stakeholder_labels = []
shipment_labels = [f"S{d.shipment_id}" for d in ethical_decisions]

for stakeholder in stakeholders:
    stakeholder_labels.append(stakeholder.name)
    utility_row = []
    for decision in ethical_decisions:
        utility_row.append(decision.stakeholder_utilities[stakeholder.stakeholder_id])
    stakeholder_matrix.append(utility_row)

im1 = ax1.imshow(stakeholder_matrix, cmap='RdYlGn', aspect='auto', vmin=0, vmax=1)
ax1.set_xticks(range(len(shipment_labels)))
ax1.set_xticklabels(shipment_labels, rotation=45)
ax1.set_yticks(range(len(stakeholder_labels)))
ax1.set_yticklabels(stakeholder_labels)
ax1.set_title('Stakeholder Utility Heatmap')
plt.colorbar(im1, ax=ax1, label='Utility Score')

# 2. Objective correlation matrix
objectives = ['Economic', 'Environmental', 'Social', 'Compliance']
objective_matrix = np.zeros((len(objectives), len(objectives)))

economic_scores = [optimizer._normalize_cost(d.economic_cost) for d in ethical_decisions]
environmental_scores = [d.environmental_impact['overall_environmental_score'] for d in ethical_decisions]
social_scores = [d.social_impact['overall_social_score'] for d in ethical_decisions]
compliance_scores = [constraint_checker.calculate_compliance_score(d.ethical_compliance) for d in ethical_decisions]

all_scores = [economic_scores, environmental_scores, social_scores, compliance_scores]

for i in range(len(objectives)):
    for j in range(len(objectives)):
        objective_matrix[i, j] = np.corrcoef(all_scores[i], all_scores[j])[0, 0]

im2 = ax2.imshow(objective_matrix, cmap='coolwarm', aspect='auto', vmin=-1, vmax=1)
ax2.set_xticks(range(len(objectives)))
ax2.set_xticklabels(objectives, rotation=45)
ax2.set_yticks(range(len(objectives)))
ax2.set_yticklabels(objectives)
ax2.set_title('Objective Correlation Matrix')
plt.colorbar(im2, ax=ax2, label='Correlation')

# Add correlation values to heatmap
for i in range(len(objectives)):
    for j in range(len(objectives)):
        text = ax2.text(j, i, f'{objective_matrix[i, j]:.2f}',
                       ha="center", va="center", color="black", fontweight='bold')

# 3. Pareto frontier analysis
ax3.scatter(economic_scores, environmental_scores, alpha=0.7, s=60, c='blue')

# Identify Pareto optimal points (simplified - points that are not dominated)
pareto_points = []
for i, (econ, env) in enumerate(zip(economic_scores, environmental_scores)):
    dominated = False
    for j, (econ_j, env_j) in enumerate(zip(economic_scores, environmental_scores)):
        if i != j and econ_j >= econ and env_j >= env and (econ_j > econ or env_j > env):
            dominated = True
            break
    if not dominated:
        pareto_points.append(i)

# Highlight Pareto optimal points
pareto_econ = [economic_scores[i] for i in pareto_points]
pareto_env = [environmental_scores[i] for i in pareto_points]
ax3.scatter(pareto_econ, pareto_env, alpha=0.9, s=100, c='red', marker='*', 
           edgecolors='black', linewidth=2, label='Pareto Optimal')

ax3.set_xlabel('Economic Performance')
ax3.set_ylabel('Environmental Performance')
ax3.set_title('Economic vs Environmental Pareto Frontier')
ax3.legend()
ax3.grid(True, alpha=0.3)
ax3.set_xlim(0, 1)
ax3.set_ylim(0, 1)

# 4. Ethical impact vs cost trade-off
total_costs = [d.economic_cost for d in ethical_decisions]
ethical_scores = [d.ethical_score for d in ethical_decisions]

# Color by chosen mode
ocean_indices = [i for i, d in enumerate(ethical_decisions) if d.chosen_mode == 'ocean']
air_indices = [i for i, d in enumerate(ethical_decisions) if d.chosen_mode == 'air']

ax4.scatter([total_costs[i] for i in ocean_indices], [ethical_scores[i] for i in ocean_indices], 
           alpha=0.7, s=60, c='blue', label='Ocean')
ax4.scatter([total_costs[i] for i in air_indices], [ethical_scores[i] for i in air_indices], 
           alpha=0.7, s=60, c='red', label='Air')

# Add trend line
z = np.polyfit(total_costs, ethical_scores, 1)
p = np.poly1d(z)
x_trend = np.linspace(min(total_costs), max(total_costs), 100)
ax4.plot(x_trend, p(x_trend), 'g--', alpha=0.8, linewidth=2, label='Trend')

ax4.set_xlabel('Total Cost ($)')
ax4.set_ylabel('Ethical Score')
ax4.set_title('Cost vs Ethical Performance Trade-Off')
ax4.legend()
ax4.grid(True, alpha=0.3)

# Add trade-off analysis annotation
cost_ethic_correlation = np.corrcoef(total_costs, ethical_scores)[0, 1]
ax4.text(0.05, 0.95, f'Cost-Ethic Correlation: {cost_ethic_correlation:.3f}',
         transform=ax4.transAxes, bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8),
         verticalalignment='top')

plt.tight_layout()
plt.show()

print("\nStakeholder Analysis Summary:")
print("=" * 80)
print(f"1. Utility Heatmap: Shows stakeholder satisfaction across all decisions")
print(f"2. Correlation Matrix: Relationships between different objectives")
print(f"3. Pareto Frontier: Economic vs environmental efficiency trade-offs")
print(f"4. Cost-Ethic Trade-Off: Relationship between cost and ethical performance")
```

### Why this Tier exists vs earlier Tiers
The Value-Aligned & Ethical Framework represents the highest level of decision-making sophistication, incorporating moral and ethical considerations into optimization:
- **Multi-stakeholder optimization**: Balances interests of customers, employees, communities, environment, and shareholders
- **Constitutional constraints**: Hard ethical requirements that cannot be violated
- **Transparency and accountability**: Full documentation of decision rationale and ethical considerations
- **Long-term sustainability**: Balances short-term economics with long-term environmental and social impacts
- **Regulatory compliance**: Ensures all decisions meet legal and ethical requirements
- **Value alignment**: Decisions reflect organizational values and ethical principles

### Pros / Cons vs Tiers 1-7
**Pros:**
- Highest ethical standards and social responsibility
- Comprehensive stakeholder satisfaction and trust building
- Regulatory compliance and risk mitigation
- Long-term sustainability and brand reputation enhancement
- Transparency for accountability and stakeholder engagement
- Alignment with corporate values and ESG requirements
- Future-proofing against evolving ethical standards and regulations

**Cons:**
- Highest implementation complexity and computational requirements
- Potential for higher costs due to ethical constraints
- Requires extensive stakeholder engagement and governance
- Complex trade-off analysis between competing objectives
- Need for continuous monitoring and compliance verification
- Potential conflicts between economic efficiency and ethical objectives
- Requires sophisticated reporting and transparency systems

### When to use this Tier
- **Publicly traded companies** with ESG reporting requirements
- **Regulated industries** with strict compliance requirements
- **Consumer-facing brands** where reputation and trust are critical
- **Supply chains** with significant environmental or social impacts
- **International operations** with diverse regulatory and cultural requirements
- **Government contracts** requiring ethical sourcing and sustainability
- **Mission-driven organizations** with strong value commitments
- **Long-term strategic planning** where sustainability is a core objective
