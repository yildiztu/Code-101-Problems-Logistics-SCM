# 38. The Automation Investment Analysis Problem
## Tier 5: The Integrated Digital Twin (System-of-Systems Simulation)

### Goal
Transform automation investment analysis into a comprehensive system-of-systems simulation that models the entire supply chain ecosystem. This approach enables real-time analysis of investment impacts across interconnected operational domains, from terminal operations to global logistics networks.

### Key assumptions
- Digital twin can accurately model physical asset behavior
- Real-time data integration provides accurate system state
- Multi-scale simulation captures system-wide impacts
- Predictive analytics can forecast investment outcomes
- What-if scenarios enable strategic decision support

### Approach (step-by-step)
1. Design multi-layer digital twin architecture
2. Implement physical asset simulation models
3. Create process simulation layer with stochastic elements
4. Build financial performance tracking system
5. Develop strategic intelligence and scenario analysis
6. Run comprehensive investment impact simulations

### What to look for in the results
- System-wide KPI improvements from automation investments
- Cascading effects across operational domains
- Network effects and ecosystem-wide value creation
- Real-time monitoring and adaptive optimization

### Concrete example (from the source)
A major container terminal implements the digital twin framework to evaluate a $75M automation portfolio spanning autonomous cranes, AI yard management, and predictive maintenance systems.

**Operational Parameters:**
- 6 berths handling 2.5M TEU annually
- 50 yard cranes and 200 truck gates
- Integration with 3 rail terminals and highway networks
- 24/7 operations with seasonal demand variations

**Digital Twin Results:**
- 23% throughput improvement
- 31% reduction in vessel waiting times
- 18% decrease in operational costs over 5 years
- Network effects creating additional $12M in annual value

### Visualization(s)
- System-of-systems architecture diagram
- Real-time KPI monitoring dashboard
- Investment impact heatmaps across domains
- Scenario comparison and sensitivity analysis

### Why this Tier exists vs earlier Tiers
This tier addresses the complexity of modern supply chain ecosystems where automation investments have cascading effects across interconnected systems. Unlike static optimization methods, the digital twin captures dynamic interactions, feedback loops, and emergent behaviors that determine real investment value.

### Pros / Cons vs earlier Tiers
**Pros vs Tier 1-4:**
- Captures system-wide impacts and network effects
- Real-time adaptation to changing conditions
- Handles complex interdependencies and feedback loops
- Enables continuous learning and improvement
- Supports strategic scenario planning

**Cons:**
- High computational requirements
- Complex implementation and maintenance
- Requires extensive data integration
- Model validation challenges
- Significant upfront investment

### When to use this Tier
- Large-scale automation investments (> $50M)
- Complex multi-system environments
- When network effects are significant
- Strategic decisions with long-term impacts
- Real-time operational optimization requirements

```python
# Import required libraries for digital twin simulation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass
from typing import List, Dict, Tuple, Optional
import random
from time import time
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Set style for better visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("Libraries imported successfully for digital twin simulation")
```

```python
@dataclass
class PhysicalAsset:
    """Represents a physical asset in the terminal"""
    id: str
    type: str  # 'crane', 'vehicle', 'gate', 'yard_equipment'
    capacity: float
    efficiency: float  # 0-1 scale
    automation_level: float  # 0-1 scale
    maintenance_cost: float
    energy_consumption: float

@dataclass
class ProcessFlow:
    """Represents a process flow in the terminal"""
    id: str
    name: str
    assets_required: List[str]
    processing_time: float
    variability: float  # Process time variability
    bottleneck_potential: float  # 0-1 scale

@dataclass
class InvestmentProject:
    """Automation investment project"""
    id: str
    name: str
    cost: float
    assets_affected: List[str]
    efficiency_improvement: float
    automation_increase: float
    implementation_time: int  # months

@dataclass
class DigitalTwinState:
    """Current state of the digital twin"""
    timestamp: datetime
    asset_states: Dict[str, Dict]
    process_performance: Dict[str, float]
    kpi_metrics: Dict[str, float]
    financial_metrics: Dict[str, float]

print("Digital twin data structures defined")
```

```python
# Initialize terminal assets and processes
def create_terminal_assets() -> List[PhysicalAsset]:
    """Create initial terminal asset configuration"""
    assets = []
    
    # Berth cranes
    for i in range(12):
        assets.append(PhysicalAsset(
            id=f"crane_{i}",
            type="crane",
            capacity=30.0,  # moves per hour
            efficiency=0.75,
            automation_level=0.2,
            maintenance_cost=5000,
            energy_consumption=150
        ))
    
    # Yard cranes
    for i in range(50):
        assets.append(PhysicalAsset(
            id=f"yard_crane_{i}",
            type="yard_crane",
            capacity=25.0,
            efficiency=0.70,
            automation_level=0.1,
            maintenance_cost=3000,
            energy_consumption=100
        ))
    
    # AGVs (Automated Guided Vehicles)
    for i in range(100):
        assets.append(PhysicalAsset(
            id=f"agv_{i}",
            type="vehicle",
            capacity=2.0,  # TEU per trip
            efficiency=0.65,
            automation_level=0.8,  # Already partially automated
            maintenance_cost=2000,
            energy_consumption=50
        ))
    
    # Truck gates
    for i in range(200):
        assets.append(PhysicalAsset(
            id=f"gate_{i}",
            type="gate",
            capacity=15.0,  # trucks per hour
            efficiency=0.80,
            automation_level=0.3,
            maintenance_cost=1000,
            energy_consumption=20
        ))
    
    return assets

def create_process_flows() -> List[ProcessFlow]:
    """Create terminal process flows"""
    flows = [
        ProcessFlow(
            id="vessel_discharge",
            name="Vessel Discharge Operations",
            assets_required=["crane_0", "agv_0", "yard_crane_0"],
            processing_time=2.5,
            variability=0.3,
            bottleneck_potential=0.8
        ),
        ProcessFlow(
            id="vessel_load",
            name="Vessel Loading Operations",
            assets_required=["crane_1", "agv_1", "yard_crane_1"],
            processing_time=2.8,
            variability=0.35,
            bottleneck_potential=0.85
        ),
        ProcessFlow(
            id="truck_gate",
            name="Truck Gate Operations",
            assets_required=["gate_0"],
            processing_time=0.5,
            variability=0.4,
            bottleneck_potential=0.6
        ),
        ProcessFlow(
            id="yard_optimization",
            name="Yard Storage Optimization",
            assets_required=["yard_crane_2"],
            processing_time=1.2,
            variability=0.25,
            bottleneck_potential=0.7
        )
    ]
    
    return flows

# Create investment projects
def create_investment_projects() -> List[InvestmentProject]:
    """Create automation investment projects"""
    projects = [
        InvestmentProject(
            id="autonomous_cranes",
            name="Autonomous Crane System",
            cost=35.0,
            assets_affected=[f"crane_{i}" for i in range(6)],  # First 6 cranes
            efficiency_improvement=0.35,
            automation_increase=0.6,
            implementation_time=18
        ),
        InvestmentProject(
            id="ai_yard_management",
            name="AI-Powered Yard Management",
            cost=25.0,
            assets_affected=[f"yard_crane_{i}" for i in range(25)],  # Half of yard cranes
            efficiency_improvement=0.40,
            automation_increase=0.7,
            implementation_time=12
        ),
        InvestmentProject(
            id="predictive_maintenance",
            name="Predictive Maintenance AI",
            cost=15.0,
            assets_affected=[f"crane_{i}" for i in range(12)] + [f"yard_crane_{i}" for i in range(50)],
            efficiency_improvement=0.15,
            automation_increase=0.4,
            implementation_time=6
        )
    ]
    
    return projects

# Initialize terminal configuration
assets = create_terminal_assets()
process_flows = create_process_flows()
investment_projects = create_investment_projects()

print(f"Digital twin initialized with:")
print(f"- {len(assets)} physical assets")
print(f"- {len(process_flows)} process flows")
print(f"- {len(investment_projects)} investment projects")
```

```python
class DigitalTwinSimulator:
    """Digital twin simulator for automation investment analysis"""
    
    def __init__(self, assets: List[PhysicalAsset], flows: List[ProcessFlow]):
        self.assets = {asset.id: asset for asset in assets}
        self.flows = {flow.id: flow for flow in flows}
        self.current_state = None
        self.simulation_history = []
        
        # Simulation parameters
        self.base_throughput = 2500000  # 2.5M TEU annually
        self.base_vessel_turnaround = 48  # hours
        self.base_operational_cost = 150000000  # $150M annually
        
    def apply_investment(self, project: InvestmentProject, completion_percentage: float):
        """Apply investment effects to assets"""
        for asset_id in project.assets_affected:
            if asset_id in self.assets:
                asset = self.assets[asset_id]
                
                # Apply efficiency improvement
                asset.efficiency += project.efficiency_improvement * completion_percentage
                asset.efficiency = min(asset.efficiency, 1.0)  # Cap at 100%
                
                # Apply automation increase
                asset.automation_level += project.automation_increase * completion_percentage
                asset.automation_level = min(asset.automation_level, 1.0)
                
                # Adjust maintenance cost (automation reduces manual maintenance)
                asset.maintenance_cost *= (1 - 0.3 * project.automation_increase * completion_percentage)
    
    def calculate_asset_performance(self, asset: PhysicalAsset) -> float:
        """Calculate individual asset performance"""
        # Performance = capacity * efficiency * (1 + automation_bonus)
        automation_bonus = asset.automation_level * 0.2  # 20% bonus at full automation
        performance = asset.capacity * asset.efficiency * (1 + automation_bonus)
        
        # Add some stochastic variation
        variation = np.random.normal(1.0, 0.05)  # 5% variation
        performance *= max(0.8, min(1.2, variation))  # Clamp variation
        
        return performance
    
    def calculate_process_performance(self, flow: ProcessFlow) -> float:
        """Calculate process flow performance"""
        # Get performance of required assets
        asset_performances = []
        for asset_id in flow.assets_required:
            # Find a representative asset of this type
            asset_type = asset_id.split('_')[0]
            matching_assets = [a for a in self.assets.values() if a.type == asset_type or asset_type in a.id]
            if matching_assets:
                avg_performance = np.mean([self.calculate_asset_performance(a) for a in matching_assets[:5]])
                asset_performances.append(avg_performance)
        
        if not asset_performances:
            return 1.0
        
        # Process performance limited by bottleneck (worst performing asset)
        bottleneck_factor = min(asset_performances)
        
        # Adjust for process variability
        variability_impact = 1 - (flow.variability * 0.1)  # Variability reduces performance
        
        return bottleneck_factor * variability_impact
    
    def calculate_system_kpis(self) -> Dict[str, float]:
        """Calculate system-wide KPIs"""
        # Calculate average asset performance
        asset_performances = [self.calculate_asset_performance(asset) for asset in self.assets.values()]
        avg_asset_performance = np.mean(asset_performances)
        
        # Calculate process performances
        process_performances = [self.calculate_process_performance(flow) for flow in self.flows.values()]
        avg_process_performance = np.mean(process_performances)
        
        # Calculate system KPIs
        kpis = {}
        
        # Throughput improvement
        throughput_multiplier = avg_asset_performance * avg_process_performance
        kpis['annual_throughput'] = self.base_throughput * throughput_multiplier
        
        # Vessel turnaround time (improvement reduces time)
        turnaround_improvement = (throughput_multiplier - 1) * 0.5  # 50% of throughput improvement affects turnaround
        kpis['vessel_turnaround_hours'] = self.base_vessel_turnaround * (1 - turnaround_improvement)
        
        # Operational costs (automation reduces costs, but initial investments add costs)
        automation_level = np.mean([asset.automation_level for asset in self.assets.values()])
        cost_reduction = automation_level * 0.15  # 15% cost reduction at full automation
        kpis['annual_operational_cost'] = self.base_operational_cost * (1 - cost_reduction)
        
        # Energy consumption
        total_energy = sum(asset.energy_consumption for asset in self.assets.values())
        energy_efficiency_improvement = automation_level * 0.25  # 25% improvement at full automation
        kpis['annual_energy_consumption'] = total_energy * 8760 * (1 - energy_efficiency_improvement)  # Annual consumption
        
        # System reliability (automation improves reliability)
        kpis['system_reliability'] = 0.85 + automation_level * 0.1  # 85% base, up to 95%
        
        return kpis
    
    def calculate_financial_metrics(self, kpis: Dict[str, float]) -> Dict[str, float]:
        """Calculate financial performance metrics"""
        metrics = {}
        
        # Revenue calculation (simplified: $50 per TEU handled)
        revenue_per_teu = 50
        metrics['annual_revenue'] = kpis['annual_throughput'] * revenue_per_teu
        
        # Operating profit
        metrics['annual_operating_profit'] = metrics['annual_revenue'] - kpis['annual_operational_cost']
        
        # Energy cost (simplified: $0.15 per kWh)
        energy_cost_per_kwh = 0.15
        metrics['annual_energy_cost'] = kpis['annual_energy_consumption'] * energy_cost_per_kwh
        
        # Maintenance cost
        metrics['annual_maintenance_cost'] = sum(asset.maintenance_cost for asset in self.assets.values()) * 12
        
        # Net profit after energy and maintenance
        metrics['annual_net_profit'] = (metrics['annual_operating_profit'] - 
                                       metrics['annual_energy_cost'] - 
                                       metrics['annual_maintenance_cost'])
        
        return metrics
    
    def simulate_step(self, timestamp: datetime) -> DigitalTwinState:
        """Simulate one time step"""
        # Calculate current system state
        kpis = self.calculate_system_kpis()
        financials = self.calculate_financial_metrics(kpis)
        
        # Calculate process performance
        process_performance = {}
        for flow_id, flow in self.flows.items():
            process_performance[flow_id] = self.calculate_process_performance(flow)
        
        # Calculate asset states
        asset_states = {}
        for asset_id, asset in self.assets.items():
            asset_states[asset_id] = {
                'performance': self.calculate_asset_performance(asset),
                'efficiency': asset.efficiency,
                'automation_level': asset.automation_level,
                'maintenance_cost': asset.maintenance_cost
            }
        
        # Create state object
        state = DigitalTwinState(
            timestamp=timestamp,
            asset_states=asset_states,
            process_performance=process_performance,
            kpi_metrics=kpis,
            financial_metrics=financials
        )
        
        self.current_state = state
        self.simulation_history.append(state)
        
        return state

print("Digital twin simulator implementation complete")
```

```python
# Initialize digital twin simulator
simulator = DigitalTwinSimulator(assets, process_flows)

# Get baseline performance (before investments)
baseline_state = simulator.simulate_step(datetime.now())

print("BASELINE PERFORMANCE (Before Automation Investments)")
print("=" * 60)
print(f"Annual Throughput: {baseline_state.kpi_metrics['annual_throughput']:,.0f} TEU")
print(f"Vessel Turnaround: {baseline_state.kpi_metrics['vessel_turnaround_hours']:.1f} hours")
print(f"Annual Operational Cost: ${baseline_state.kpi_metrics['annual_operational_cost']:,.0f}")
print(f"System Reliability: {baseline_state.kpi_metrics['system_reliability']:.1%}")
print(f"Annual Revenue: ${baseline_state.financial_metrics['annual_revenue']:,.0f}")
print(f"Annual Net Profit: ${baseline_state.financial_metrics['annual_net_profit']:,.0f}")

# Calculate baseline asset performance
baseline_automation = np.mean([asset.automation_level for asset in simulator.assets.values()])
baseline_efficiency = np.mean([asset.efficiency for asset in simulator.assets.values()])

print(f"\nBaseline Asset Metrics:")
print(f"Average Automation Level: {baseline_automation:.1%}")
print(f"Average Efficiency: {baseline_efficiency:.1%}")

# Store baseline for comparison
baseline_kpis = baseline_state.kpi_metrics.copy()
baseline_financials = baseline_state.financial_metrics.copy()
```

```python
def simulate_investment_implementation(simulator: DigitalTwinSimulator, 
                                     projects: List[InvestmentProject],
                                     simulation_months: int = 60) -> List[DigitalTwinState]:
    """Simulate investment implementation over time"""
    
    simulation_results = []
    start_time = datetime.now()
    
    # Project implementation schedule
    project_schedule = {}
    for project in projects:
        # Stagger project start times
        start_month = projects.index(project) * 6  # Start every 6 months
        project_schedule[project.id] = {
            'start_month': start_month,
            'end_month': start_month + project.implementation_time,
            'project': project
        }
    
    print(f"Simulating {len(projects)} projects over {simulation_months} months...")
    print(f"Total Investment: ${sum(p.cost for p in projects):.1f}M")
    
    for month in range(simulation_months + 1):
        current_time = start_time + timedelta(days=30 * month)
        
        # Apply investment progress for active projects
        for project_id, schedule in project_schedule.items():
            if schedule['start_month'] <= month <= schedule['end_month']:
                project = schedule['project']
                progress = (month - schedule['start_month']) / project.implementation_time
                progress = min(1.0, progress)
                
                # Apply investment effects
                simulator.apply_investment(project, progress)
        
        # Simulate current state
        state = simulator.simulate_step(current_time)
        simulation_results.append(state)
        
        # Progress reporting
        if month % 12 == 0:
            year = month // 12
            print(f"Year {year}: Throughput = {state.kpi_metrics['annual_throughput']:,.0f} TEU, "
                  f"Net Profit = ${state.financial_metrics['annual_net_profit']:,.0f}")
    
    return simulation_results

# Run investment simulation
simulation_results = simulate_investment_implementation(
    simulator, investment_projects, simulation_months=60
)

print("\n" + "=" * 60)
print("INVESTMENT SIMULATION RESULTS")
print("=" * 60)

# Get final state
final_state = simulation_results[-1]

print(f"\nFinal Performance (After 5 Years):")
print(f"Annual Throughput: {final_state.kpi_metrics['annual_throughput']:,.0f} TEU")
print(f"Vessel Turnaround: {final_state.kpi_metrics['vessel_turnaround_hours']:.1f} hours")
print(f"Annual Operational Cost: ${final_state.kpi_metrics['annual_operational_cost']:,.0f}")
print(f"System Reliability: {final_state.kpi_metrics['system_reliability']:.1%}")
print(f"Annual Revenue: ${final_state.financial_metrics['annual_revenue']:,.0f}")
print(f"Annual Net Profit: ${final_state.financial_metrics['annual_net_profit']:,.0f}")

# Calculate improvements
throughput_improvement = (final_state.kpi_metrics['annual_throughput'] - baseline_kpis['annual_throughput']) / baseline_kpis['annual_throughput'] * 100
turnaround_improvement = (baseline_kpis['vessel_turnaround_hours'] - final_state.kpi_metrics['vessel_turnaround_hours']) / baseline_kpis['vessel_turnaround_hours'] * 100
cost_reduction = (baseline_kpis['annual_operational_cost'] - final_state.kpi_metrics['annual_operational_cost']) / baseline_kpis['annual_operational_cost'] * 100
profit_improvement = (final_state.financial_metrics['annual_net_profit'] - baseline_financials['annual_net_profit']) / abs(baseline_financials['annual_net_profit']) * 100

print(f"\nKey Improvements:")
print(f"- Throughput Improvement: {throughput_improvement:.1f}%")
print(f"- Turnaround Time Reduction: {turnaround_improvement:.1f}%")
print(f"- Operational Cost Reduction: {cost_reduction:.1f}%")
print(f"- Net Profit Improvement: {profit_improvement:.1f}%")

# Calculate network effects (additional value beyond direct improvements)
direct_profit_increase = final_state.financial_metrics['annual_net_profit'] - baseline_financials['annual_net_profit']
network_effect_value = direct_profit_increase * 0.15  # Estimated 15% network effect

print(f"\nNetwork Effects Analysis:")
print(f"- Direct Profit Increase: ${direct_profit_increase:,.0f}")
print(f"- Estimated Network Effect Value: ${network_effect_value:,.0f}")
print(f"- Total Value Creation: ${direct_profit_increase + network_effect_value:,.0f}")
```

```python
# Asset performance analysis and network effects
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))

# Analyze asset performance by type
asset_types = {}
for asset in simulator.assets.values():
    if asset.type not in asset_types:
        asset_types[asset.type] = []
    asset_types[asset.type].append(asset)

# Calculate performance metrics by asset type
type_performance = {}
for asset_type, type_assets in asset_types.items():
    performances = [simulator.calculate_asset_performance(asset) for asset in type_assets]
    automation_levels = [asset.automation_level for asset in type_assets]
    efficiencies = [asset.efficiency for asset in type_assets]
    
    type_performance[asset_type] = {
        'avg_performance': np.mean(performances),
        'avg_automation': np.mean(automation_levels),
        'avg_efficiency': np.mean(efficiencies),
        'count': len(type_assets)
    }

# Plot 1: Asset performance by type
type_names = list(type_performance.keys())
performances = [type_performance[t]['avg_performance'] for t in type_names]
colors = plt.cm.Set3(np.linspace(0, 1, len(type_names)))

bars1 = ax1.bar(type_names, performances, color=colors, alpha=0.8)
ax1.set_ylabel('Average Performance')
ax1.set_title('Asset Performance by Type')
ax1.grid(True, alpha=0.3)
plt.setp(ax1.get_xticklabels(), rotation=45, ha='right')

# Add value labels on bars
for bar, perf in zip(bars1, performances):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.5,
             f'{perf:.1f}', ha='center', va='bottom')

# Plot 2: Automation levels by asset type
automation_levels = [type_performance[t]['avg_automation'] * 100 for t in type_names]
efficiency_levels = [type_performance[t]['avg_efficiency'] * 100 for t in type_names]

x = np.arange(len(type_names))
width = 0.35

bars2 = ax2.bar(x - width/2, automation_levels, width, label='Automation Level', alpha=0.8)
bars3 = ax2.bar(x + width/2, efficiency_levels, width, label='Efficiency', alpha=0.8)

ax2.set_ylabel('Percentage (%)')
ax2.set_title('Automation vs Efficiency by Asset Type')
ax2.set_xticks(x)
ax2.set_xticklabels(type_names, rotation=45, ha='right')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Plot 3: Process performance heatmap
process_names = list(simulator.flows.keys())
process_perf_values = [final_state.process_performance[pid] for pid in process_names]

# Create heatmap data
heatmap_data = np.array(process_perf_values).reshape(-1, 1)

im = ax3.imshow(heatmap_data, cmap='RdYlGn', aspect='auto', vmin=0.5, vmax=1.5)
ax3.set_xticks([0])
ax3.set_xticklabels(['Performance'])
ax3.set_yticks(range(len(process_names)))
ax3.set_yticklabels([name.replace('_', ' ').title() for name in process_names])
ax3.set_title('Process Performance Heatmap')

# Add performance values
for i, value in enumerate(process_perf_values):
    ax3.text(0, i, f'{value:.2f}', ha='center', va='center', fontweight='bold')

# Plot 4: Investment ROI analysis
investment_costs = [p.cost for p in investment_projects]
project_names = [p.name.replace(' ', '\n') for p in investment_projects]

# Calculate individual project ROI (simplified)
project_rois = []
for i, project in enumerate(investment_projects):
    # Estimate project contribution to total improvement
    project_contribution = project.cost / sum(investment_costs) * profit_improvement
    annual_return = (baseline_financials['annual_net_profit'] * project_contribution / 100)
    roi = (annual_return * 5) / project.cost * 100  # 5-year ROI
    project_rois.append(roi)

# Create dual-axis plot
ax4_twin = ax4.twinx()

bars4 = ax4.bar(project_names, investment_costs, alpha=0.7, color='orange', label='Investment Cost')
line4 = ax4_twin.plot(project_names, project_rois, 'bo-', linewidth=3, markersize=8, label='5-Year ROI')

ax4.set_ylabel('Investment Cost (Million $)', color='orange')
ax4_twin.set_ylabel('5-Year ROI (%)', color='blue')
ax4.set_title('Investment Cost vs ROI Analysis')
plt.setp(ax4.get_xticklabels(), rotation=45, ha='right')
ax4.grid(True, alpha=0.3)

# Add ROI labels
for i, roi in enumerate(project_rois):
    ax4_twin.text(i, roi + 1, f'{roi:.0f}%', ha='center', va='bottom', color='blue')

plt.tight_layout()
plt.show()

print("Asset performance and investment analysis visualized")
```

```python
# Scenario analysis and what-if simulations
print("\n" + "=" * 80)
print("SCENARIO ANALYSIS AND WHAT-IF SIMULATIONS")
print("=" * 80)

def run_scenario_analysis(base_simulator: DigitalTwinSimulator, 
                         base_projects: List[InvestmentProject]) -> Dict[str, Dict]:
    """Run multiple what-if scenarios"""
    
    scenarios = {}
    
    # Scenario 1: Baseline (current plan)
    simulator1 = DigitalTwinSimulator(assets, process_flows)
    results1 = simulate_investment_implementation(simulator1, base_projects, simulation_months=60)
    scenarios['baseline'] = {
        'final_state': results1[-1],
        'total_investment': sum(p.cost for p in base_projects)
    }
    
    # Scenario 2: Accelerated implementation
    accelerated_projects = []
    for project in base_projects:
        accelerated_project = InvestmentProject(
            id=project.id,
            name=project.name,
            cost=project.cost * 1.2,  # 20% cost premium for speed
            assets_affected=project.assets_affected,
            efficiency_improvement=project.efficiency_improvement,
            automation_increase=project.automation_increase,
            implementation_time=project.implementation_time // 2  # Half the time
        )
        accelerated_projects.append(accelerated_project)
    
    simulator2 = DigitalTwinSimulator(assets, process_flows)
    results2 = simulate_investment_implementation(simulator2, accelerated_projects, simulation_months=36)
    scenarios['accelerated'] = {
        'final_state': results2[-1],
        'total_investment': sum(p.cost for p in accelerated_projects)
    }
    
    # Scenario 3: Phased implementation (only 2 projects initially)
    phased_projects = base_projects[:2]  # Only first two projects
    simulator3 = DigitalTwinSimulator(assets, process_flows)
    results3 = simulate_investment_implementation(simulator3, phased_projects, simulation_months=60)
    scenarios['phased'] = {
        'final_state': results3[-1],
        'total_investment': sum(p.cost for p in phased_projects)
    }
    
    # Scenario 4: Full automation (add additional project)
    full_projects = base_projects.copy()
    full_projects.append(InvestmentProject(
        id="full_gate_automation",
        name="Full Gate Automation",
        cost=20.0,
        assets_affected=[f"gate_{i}" for i in range(100)],  # Half of gates
        efficiency_improvement=0.50,
        automation_increase=0.8,
        implementation_time=24
    ))
    
    simulator4 = DigitalTwinSimulator(assets, process_flows)
    results4 = simulate_investment_implementation(simulator4, full_projects, simulation_months=72)
    scenarios['full_automation'] = {
        'final_state': results4[-1],
        'total_investment': sum(p.cost for p in full_projects)
    }
    
    return scenarios

# Run scenario analysis
scenarios = run_scenario_analysis(simulator, investment_projects)

print("\nScenario Comparison Results:")
print("=" * 50)

for scenario_name, scenario_data in scenarios.items():
    state = scenario_data['final_state']
    investment = scenario_data['total_investment']
    
    print(f"\n{scenario_name.title()} Scenario:")
    print(f"  Total Investment: ${investment:.1f}M")
    print(f"  Annual Throughput: {state.kpi_metrics['annual_throughput']:,.0f} TEU")
    print(f"  Net Profit: ${state.financial_metrics['annual_net_profit']:,.0f}")
    print(f"  System Reliability: {state.kpi_metrics['system_reliability']:.1%}")
    
    # Calculate ROI
    profit_increase = state.financial_metrics['annual_net_profit'] - baseline_financials['annual_net_profit']
    roi_5year = (profit_increase * 5) / investment * 100
    print(f"  5-Year ROI: {roi_5year:.1f}%")

# Create scenario comparison visualization
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))

scenario_names = list(scenarios.keys())
throughputs = [scenarios[s]['final_state'].kpi_metrics['annual_throughput']/1e6 for s in scenario_names]
profits = [scenarios[s]['final_state'].financial_metrics['annual_net_profit']/1e6 for s in scenario_names]
investments = [scenarios[s]['total_investment'] for s in scenario_names]
reliabilities = [scenarios[s]['final_state'].kpi_metrics['system_reliability']*100 for s in scenario_names]

# Plot 1: Throughput comparison
bars1 = ax1.bar(scenario_names, throughputs, color='skyblue', alpha=0.8)
ax1.axhline(y=baseline_kpis['annual_throughput']/1e6, color='red', linestyle='--', alpha=0.7, label='Baseline')
ax1.set_ylabel('Annual Throughput (Million TEU)')
ax1.set_title('Throughput Comparison Across Scenarios')
ax1.legend()
ax1.grid(True, alpha=0.3)
plt.setp(ax1.get_xticklabels(), rotation=45, ha='right')

# Plot 2: Profit comparison
bars2 = ax2.bar(scenario_names, profits, color='lightgreen', alpha=0.8)
ax2.axhline(y=baseline_financials['annual_net_profit']/1e6, color='red', linestyle='--', alpha=0.7, label='Baseline')
ax2.set_ylabel('Annual Net Profit (Million $)')
ax2.set_title('Profit Comparison Across Scenarios')
ax2.legend()
ax2.grid(True, alpha=0.3)
plt.setp(ax2.get_xticklabels(), rotation=45, ha='right')

# Plot 3: Investment vs ROI
ax3_twin = ax3.twinx()
bars3 = ax3.bar(scenario_names, investments, alpha=0.7, color='orange', label='Investment')
rois = [(scenarios[s]['final_state'].financial_metrics['annual_net_profit'] - baseline_financials['annual_net_profit']) * 5 / scenarios[s]['total_investment'] * 100 for s in scenario_names]
line3 = ax3_twin.plot(scenario_names, rois, 'ro-', linewidth=3, markersize=8, label='5-Year ROI')

ax3.set_ylabel('Total Investment (Million $)', color='orange')
ax3_twin.set_ylabel('5-Year ROI (%)', color='red')
ax3.set_title('Investment vs ROI Comparison')
ax3.grid(True, alpha=0.3)
plt.setp(ax3.get_xticklabels(), rotation=45, ha='right')

# Plot 4: Radar chart for scenario comparison
categories = ['Throughput', 'Profit', 'Reliability', 'Cost Efficiency']
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]  # Complete the circle

for i, scenario_name in enumerate(scenario_names):
    values = [
        throughputs[i] / max(throughputs) * 100,  # Normalized throughput
        profits[i] / max(profits) * 100,  # Normalized profit
        reliabilities[i],  # Already in percentage
        (max(investments) - investments[i]) / max(investments) * 100  # Inverted cost (lower is better)
    ]
    values += values[:1]  # Complete the circle
    
    ax4.plot(angles, values, 'o-', linewidth=2, label=scenario_name.title())
    ax4.fill(angles, values, alpha=0.25)

ax4.set_xticks(angles[:-1])
ax4.set_xticklabels(categories)
ax4.set_ylim([0, 100])
ax4.set_title('Scenario Performance Radar')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "=" * 80)
print("DIGITAL TWIN ANALYSIS COMPLETE")
print("=" * 80)
```
