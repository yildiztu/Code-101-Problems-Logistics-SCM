# 14. The Programmable Matter Container Stowage Problem

## Problem Introduction

This notebook explores **Human-AI Partnership** models for complex container stowage decisions in large-scale container terminals. The system combines human expertise with AI capabilities to handle scenarios that are too complex for pure automation while maintaining operational efficiency and safety.

### Partnership Scope:
- **Vessel Scale**: 22,000 TEU ultra-large container vessel
- **Route Complexity**: 12-port global rotation with diverse cargo types
- **Human Factors**: Expert stowage coordinators with domain knowledge
- **AI Capabilities**: Machine learning models for optimization and prediction
- **Decision Framework**: Collaborative decision-making with confidence scoring

The Human-AI Partnership approach leverages the strengths of both human experts (experience, intuition, contextual understanding) and AI systems (computational power, pattern recognition, optimization) to achieve superior stowage outcomes.

## 1. Human-AI Partnership Framework

### Partnership Architecture

The Human-AI Partnership system consists of several key components that work together to support collaborative decision-making:

#### **Human Expert Layer**:
- Senior stowage coordinators with 10+ years experience
- Domain knowledge in cargo handling and safety regulations
- Intuitive understanding of complex operational constraints
- Ability to handle unprecedented situations

#### **AI Assistant Layer**:
- Machine learning models for pattern recognition
- Optimization algorithms for candidate solutions
- Risk assessment and prediction models
- Real-time data processing and analysis

#### **Collaboration Interface**:
- Interactive decision support dashboard
- Confidence scoring and recommendation system
- Scenario comparison and what-if analysis
- Learning and adaptation mechanisms

Let's implement the partnership framework:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
from typing import Dict, List, Tuple, Optional, Any
from dataclasses import dataclass, field
from enum import Enum
import random
from collections import defaultdict
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)

class DecisionLevel(Enum):
    """Levels of human-AI collaboration"""
    AI_AUTONOMOUS = "AI_Autonomous"      # AI makes decision
    AI_RECOMMENDS = "AI_Recommends"      # AI recommends, human decides
    HUMAN_LEADS = "Human_Leads"          # Human leads, AI assists
    HUMAN_AUTONOMOUS = "Human_Autonomous"  # Human makes decision

class ConfidenceLevel(Enum):
    """Confidence levels in recommendations"""
    VERY_LOW = 1
    LOW = 2
    MEDIUM = 3
    HIGH = 4
    VERY_HIGH = 5

@dataclass
class HumanExpert:
    """Represents a human stowage expert"""
    expert_id: str
    name: str
    experience_years: int
    specializations: List[str]
    confidence_calibration: float  # How well-calibrated their confidence is
    risk_tolerance: float  # 0-1, risk tolerance level
    decision_speed: float  # Average decision time in minutes
    
    def assess_scenario(self, scenario: Dict) -> Tuple[float, str]:
        """Human expert assesses a stowage scenario"""
        # Simulate human assessment based on experience
        base_confidence = min(0.9, 0.5 + self.experience_years * 0.03)
        
        # Adjust for specializations
        specialization_bonus = 0.0
        for spec in self.specializations:
            if spec in scenario.get('complexity_factors', []):
                specialization_bonus += 0.1
        
        # Add some realistic variation
        confidence = min(0.95, base_confidence + specialization_bonus + np.random.normal(0, 0.1))
        confidence = max(0.1, confidence)
        
        # Generate reasoning
        reasoning = self._generate_reasoning(scenario)
        
        return confidence, reasoning
    
    def _generate_reasoning(self, scenario: Dict) -> str:
        """Generate human-like reasoning for the assessment"""
        factors = scenario.get('complexity_factors', [])
        
        if 'hazardous_cargo' in factors:
            return "Based on IMDG code requirements, we need to ensure proper segregation of hazardous materials. I recommend placing Class 1 explosives in isolated fore bay with 3-meter clearance."
        elif 'reefer_density' in factors:
            return "High reefer concentration requires careful power management. We should cluster reefer containers near power stations but maintain airflow access."
        elif 'weight_distribution' in factors:
            return "The weight distribution is critical for vessel stability. We need to ensure proper longitudinal and transverse balance to meet SOLAS requirements."
        elif 'port_congestion' in factors:
            return "Given port congestion at the next destination, we should prioritize containers for early discharge to minimize dwell time."
        else:
            return "Standard stowage principles apply. We should focus on minimizing overstowage while maintaining vessel stability."

@dataclass
class AIAssistant:
    """Represents an AI assistant for stowage decisions"""
    model_version: str
    training_data_size: int
    accuracy_metrics: Dict[str, float]
    processing_speed: float  # Average processing time in seconds
    confidence_calibration: float
    
    def analyze_scenario(self, scenario: Dict) -> Tuple[float, str, Dict]:
        """AI assistant analyzes a stowage scenario"""
        
        # Simulate AI analysis based on training and scenario complexity
        base_confidence = self.accuracy_metrics.get('overall_accuracy', 0.85)
        
        # Adjust for scenario complexity
        complexity_penalty = 0.0
        if scenario.get('complexity_score', 0) > 0.7:
            complexity_penalty = 0.15
        elif scenario.get('novelty_score', 0) > 0.5:
            complexity_penalty = 0.25
        
        confidence = max(0.2, base_confidence - complexity_penalty)
        
        # Generate AI reasoning
        reasoning = self._generate_ai_reasoning(scenario)
        
        # Generate detailed analysis
        analysis = self._generate_detailed_analysis(scenario)
        
        return confidence, reasoning, analysis
    
    def _generate_ai_reasoning(self, scenario: Dict) -> str:
        """Generate AI reasoning based on pattern matching"""
        factors = scenario.get('complexity_factors', [])
        
        if 'hazardous_cargo' in factors:
            return "Analysis of 15,000 historical hazardous cargo stowage cases indicates optimal segregation pattern: fore bay isolation with 3.2m average clearance reduces risk by 73%."
        elif 'reefer_density' in factors:
            return "Power consumption modeling suggests clustering reefer containers within 2 bays of power stations reduces cable length by 45% and improves energy efficiency by 12%."
        elif 'weight_distribution' in factors:
            return "Stability simulation shows optimal LCG at 122.5m with weight distribution coefficient of 0.82. Current configuration achieves 94% of optimal stability metric."
        elif 'port_congestion' in factors:
            return "Predictive analysis of port operations indicates 67% probability of congestion. Recommending prioritized discharge sequence reduces expected dwell time by 4.2 hours."
        else:
            return "Pattern matching against 50,000 historical cases suggests standard overstowage minimization strategy with 89% success probability."
    
    def _generate_detailed_analysis(self, scenario: Dict) -> Dict:
        """Generate detailed AI analysis"""
        return {
            'risk_assessment': {
                'stability_risk': np.random.uniform(0.1, 0.3),
                'handling_risk': np.random.uniform(0.05, 0.2),
                'safety_risk': np.random.uniform(0.02, 0.15)
            },
            'optimization_metrics': {
                'overstowage_reduction': np.random.uniform(15, 35),
                'handling_efficiency': np.random.uniform(0.8, 0.95),
                'fuel_efficiency': np.random.uniform(0.02, 0.08)
            },
            'confidence_breakdown': {
                'pattern_similarity': np.random.uniform(0.7, 0.9),
                'data_quality': np.random.uniform(0.8, 0.95),
                'model_uncertainty': np.random.uniform(0.1, 0.3)
            }
        }

print("Human-AI Partnership framework initialized")
```

### Partnership Decision Engine

The core of the system is the decision engine that determines how humans and AI should collaborate based on the scenario characteristics and confidence levels.

```python
@dataclass
class StowageScenario:
    """Represents a complex stowage scenario"""
    scenario_id: str
    timestamp: datetime
    vessel_id: str
    port_sequence: List[str]
    cargo_manifest: List[Dict]
    complexity_factors: List[str]
    complexity_score: float  # 0-1
    novelty_score: float     # 0-1
    time_pressure: float     # 0-1
    safety_critical: bool
    
@dataclass
class PartnershipDecision:
    """Represents a collaborative decision outcome"""
    scenario_id: str
    decision_level: DecisionLevel
    final_confidence: float
    human_confidence: float
    ai_confidence: float
    decision_time: float  # minutes
    final_recommendation: str
    reasoning: str
    consensus_score: float  # 0-1, agreement between human and AI

class PartnershipDecisionEngine:
    """Engine for managing human-AI partnership decisions"""
    
    def __init__(self):
        self.human_experts = self._initialize_human_experts()
        self.ai_assistant = self._initialize_ai_assistant()
        self.decision_history = []
        self.performance_metrics = defaultdict(list)
    
    def _initialize_human_experts(self) -> List[HumanExpert]:
        """Initialize team of human experts"""
        return [
            HumanExpert(
                expert_id="EXP001",
                name="Sarah Chen",
                experience_years=15,
                specializations=["hazardous_cargo", "weight_distribution"],
                confidence_calibration=0.85,
                risk_tolerance=0.3,
                decision_speed=8.5
            ),
            HumanExpert(
                expert_id="EXP002",
                name="Michael Rodriguez",
                experience_years=12,
                specializations=["reefer_operations", "port_congestion"],
                confidence_calibration=0.78,
                risk_tolerance=0.4,
                decision_speed=6.2
            ),
            HumanExpert(
                expert_id="EXP003",
                name="Aisha Patel",
                experience_years=18,
                specializations=["stability", "hazardous_cargo"],
                confidence_calibration=0.92,
                risk_tolerance=0.25,
                decision_speed=10.1
            )
        ]
    
    def _initialize_ai_assistant(self) -> AIAssistant:
        """Initialize AI assistant"""
        return AIAssistant(
            model_version="StowageAI-v3.2",
            training_data_size=50000,
            accuracy_metrics={
                'overall_accuracy': 0.87,
                'hazardous_cargo': 0.91,
                'reefer_operations': 0.84,
                'stability': 0.89,
                'port_congestion': 0.82
            },
            processing_speed=2.3,
            confidence_calibration=0.81
        )
    
    def process_scenario(self, scenario: StowageScenario) -> PartnershipDecision:
        """Process a stowage scenario through human-AI partnership"""
        
        # Select best human expert for this scenario
        selected_expert = self._select_expert(scenario)
        
        # Get human assessment
        human_confidence, human_reasoning = selected_expert.assess_scenario(scenario)
        
        # Get AI analysis
        ai_confidence, ai_reasoning, ai_analysis = self.ai_assistant.analyze_scenario(scenario)
        
        # Determine optimal collaboration level
        decision_level = self._determine_collaboration_level(
            scenario, human_confidence, ai_confidence, selected_expert
        )
        
        # Generate collaborative decision
        final_decision = self._generate_collaborative_decision(
            scenario, decision_level, human_confidence, ai_confidence,
            human_reasoning, ai_reasoning, selected_expert
        )
        
        # Record decision
        self.decision_history.append(final_decision)
        
        return final_decision
    
    def _select_expert(self, scenario: StowageScenario) -> HumanExpert:
        """Select the best human expert for the scenario"""
        best_expert = None
        best_score = -1
        
        for expert in self.human_experts:
            # Calculate expertise match score
            match_score = 0.0
            for factor in scenario.complexity_factors:
                if factor in expert.specializations:
                    match_score += 1.0
            
            # Normalize by number of factors
            if scenario.complexity_factors:
                match_score /= len(scenario.complexity_factors)
            
            # Add experience factor
            experience_factor = expert.experience_years / 20.0  # Normalize to 0-1
            total_score = match_score * 0.7 + experience_factor * 0.3
            
            if total_score > best_score:
                best_score = total_score
                best_expert = expert
        
        return best_expert or self.human_experts[0]
    
    def _determine_collaboration_level(self, scenario: StowageScenario,
                                     human_confidence: float, ai_confidence: float,
                                     expert: HumanExpert) -> DecisionLevel:
        """Determine the optimal level of human-AI collaboration"""
        
        # High-stakes scenarios require human oversight
        if scenario.safety_critical or scenario.complexity_score > 0.8:
            if human_confidence > 0.8:
                return DecisionLevel.HUMAN_LEADS
            else:
                return DecisionLevel.AI_RECOMMENDS
        
        # Time pressure scenarios
        if scenario.time_pressure > 0.7:
            if ai_confidence > 0.8 and expert.decision_speed > 5.0:
                return DecisionLevel.AI_AUTONOMOUS
            else:
                return DecisionLevel.AI_RECOMMENDS
        
        # Novel scenarios
        if scenario.novelty_score > 0.6:
            return DecisionLevel.HUMAN_LEADS
        
        # Standard scenarios with high AI confidence
        if ai_confidence > 0.9 and human_confidence > 0.7:
            return DecisionLevel.AI_AUTONOMOUS
        
        # Default collaborative approach
        return DecisionLevel.AI_RECOMMENDS
    
    def _generate_collaborative_decision(self, scenario: StowageScenario,
                                       decision_level: DecisionLevel,
                                       human_confidence: float, ai_confidence: float,
                                       human_reasoning: str, ai_reasoning: str,
                                       expert: HumanExpert) -> PartnershipDecision:
        """Generate the final collaborative decision"""
        
        # Calculate decision time based on collaboration level
        if decision_level == DecisionLevel.AI_AUTONOMOUS:
            decision_time = self.ai_assistant.processing_speed / 60.0  # Convert to minutes
            final_confidence = ai_confidence
            final_reasoning = ai_reasoning
            consensus_score = 1.0  # AI-only decision
        
        elif decision_level == DecisionLevel.AI_RECOMMENDS:
            decision_time = self.ai_assistant.processing_speed / 60.0 + expert.decision_speed * 0.5
            final_confidence = (ai_confidence + human_confidence) / 2.0
            final_reasoning = f"AI Recommendation: {ai_reasoning}\n\nHuman Assessment: {human_reasoning}"
            consensus_score = 1.0 - abs(ai_confidence - human_confidence)
        
        elif decision_level == DecisionLevel.HUMAN_LEADS:
            decision_time = expert.decision_speed * 0.8 + self.ai_assistant.processing_speed / 60.0 * 0.3
            final_confidence = human_confidence * 0.7 + ai_confidence * 0.3
            final_reasoning = f"Human Decision: {human_reasoning}\n\nAI Support: {ai_reasoning}"
            consensus_score = 1.0 - abs(ai_confidence - human_confidence)
        
        else:  # HUMAN_AUTONOMOUS
            decision_time = expert.decision_speed
            final_confidence = human_confidence
            final_reasoning = human_reasoning
            consensus_score = 1.0  # Human-only decision
        
        # Generate final recommendation
        final_recommendation = self._generate_recommendation(scenario, decision_level, final_confidence)
        
        return PartnershipDecision(
            scenario_id=scenario.scenario_id,
            decision_level=decision_level,
            final_confidence=final_confidence,
            human_confidence=human_confidence,
            ai_confidence=ai_confidence,
            decision_time=decision_time,
            final_recommendation=final_recommendation,
            reasoning=final_reasoning,
            consensus_score=consensus_score
        )
    
    def _generate_recommendation(self, scenario: StowageScenario,
                                decision_level: DecisionLevel,
                                confidence: float) -> str:
        """Generate final stowage recommendation"""
        
        if confidence > 0.8:
            confidence_str = "High confidence"
        elif confidence > 0.6:
            confidence_str = "Moderate confidence"
        else:
            confidence_str = "Low confidence"
        
        if 'hazardous_cargo' in scenario.complexity_factors:
            return f"{confidence_str}: Implement hazardous cargo segregation plan with fore bay isolation and 3-meter clearance zones."
        elif 'reefer_density' in scenario.complexity_factors:
            return f"{confidence_str}: Cluster reefer containers near power stations with optimized cable routing and airflow management."
        elif 'weight_distribution' in scenario.complexity_factors:
            return f"{confidence_str}: Optimize weight distribution with target LCG at 122.5m and stability coefficient above 0.8."
        elif 'port_congestion' in scenario.complexity_factors:
            return f"{confidence_str}: Prioritize discharge sequence for congested ports with expected 4+ hour dwell time reduction."
        else:
            return f"{confidence_str}: Implement standard overstowage minimization strategy with stability optimization."

# Initialize partnership decision engine
partnership_engine = PartnershipDecisionEngine()
print("Partnership decision engine initialized")
print(f"Human experts: {len(partnership_engine.human_experts)}")
print(f"AI assistant: {partnership_engine.ai_assistant.model_version}")
```

## 2. Complex Scenario Generation

Let's generate realistic complex stowage scenarios that would benefit from human-AI collaboration.

```python
class ScenarioGenerator:
    """Generates complex stowage scenarios for testing"""
    
    def __init__(self):
        self.global_routes = [
            ['Singapore', 'Colombo', 'Dubai', 'Suez', 'Rotterdam', 'Hamburg', 'Antwerp', 'New York'],
            ['Shanghai', 'Ningbo', 'Busan', 'Tokyo', 'Los Angeles', 'Long Beach', 'Panama', 'New York'],
            ['Hong Kong', 'Shenzhen', 'Singapore', 'Jebel Ali', 'Suez', 'Genoa', 'Barcelona', 'Southampton']
        ]
        
        self.cargo_types = {
            'standard': {'weight_range': (10, 25), 'probability': 0.6},
            'reefer': {'weight_range': (15, 30), 'probability': 0.15},
            'hazardous': {'weight_range': (8, 20), 'probability': 0.08},
            'oversized': {'weight_range': (25, 40), 'probability': 0.07},
            'high_value': {'weight_range': (5, 15), 'probability': 0.1}
        }
    
    def generate_complex_scenario(self, scenario_id: str) -> StowageScenario:
        """Generate a complex stowage scenario"""
        
        # Select route
        route = np.random.choice(self.global_routes)
        
        # Generate complexity factors
        possible_factors = [
            'hazardous_cargo', 'reefer_density', 'weight_distribution', 
            'port_congestion', 'time_constraints', 'weather_risk',
            'stability_concerns', 'customs_issues', 'equipment_limitations'
        ]
        
        # Select 2-4 complexity factors
        num_factors = np.random.randint(2, 5)
        complexity_factors = np.random.choice(possible_factors, num_factors, replace=False).tolist()
        
        # Calculate complexity and novelty scores
        complexity_score = min(1.0, num_factors * 0.2 + np.random.uniform(0, 0.3))
        novelty_score = np.random.uniform(0.3, 0.8) if 'customs_issues' in complexity_factors else np.random.uniform(0.1, 0.5)
        
        # Time pressure based on factors
        time_pressure = 0.8 if 'time_constraints' in complexity_factors else np.random.uniform(0.2, 0.6)
        
        # Safety critical if hazardous or stability concerns
        safety_critical = any(factor in ['hazardous_cargo', 'stability_concerns'] for factor in complexity_factors)
        
        # Generate cargo manifest
        cargo_manifest = self._generate_cargo_manifest(complexity_factors)
        
        return StowageScenario(
            scenario_id=scenario_id,
            timestamp=datetime.now(),
            vessel_id="ULCS-22000-01",
            port_sequence=route,
            cargo_manifest=cargo_manifest,
            complexity_factors=complexity_factors,
            complexity_score=complexity_score,
            novelty_score=novelty_score,
            time_pressure=time_pressure,
            safety_critical=safety_critical
        )
    
    def _generate_cargo_manifest(self, complexity_factors: List[str]) -> List[Dict]:
        """Generate cargo manifest based on complexity factors"""
        manifest = []
        num_containers = np.random.randint(8000, 12000)  # Large vessel
        
        # Adjust cargo distribution based on complexity factors
        cargo_adjustments = {}
        if 'hazardous_cargo' in complexity_factors:
            cargo_adjustments['hazardous'] = 0.15  # Increased hazardous cargo
        if 'reefer_density' in complexity_factors:
            cargo_adjustments['reefer'] = 0.25   # Increased reefer density
        if 'weight_distribution' in complexity_factors:
            cargo_adjustments['oversized'] = 0.12  # More oversized cargo
        
        for i in range(num_containers):
            # Select cargo type based on probabilities
            cargo_type = self._select_cargo_type(cargo_adjustments)
            
            weight_range = self.cargo_types[cargo_type]['weight_range']
            weight = np.random.uniform(*weight_range)
            
            destination = np.random.choice(['Singapore', 'Dubai', 'Rotterdam', 'New York', 'Los Angeles'])
            
            manifest.append({
                'container_id': f"CONT{i:06d}",
                'type': cargo_type,
                'weight': weight,
                'destination': destination,
                'special_requirements': self._get_special_requirements(cargo_type, complexity_factors)
            })
        
        return manifest
    
    def _select_cargo_type(self, adjustments: Dict[str, float]) -> str:
        """Select cargo type with adjustments for complexity factors"""
        # Base probabilities
        probs = {cargo_type: info['probability'] for cargo_type, info in self.cargo_types.items()}
        
        # Apply adjustments
        for cargo_type, adjustment in adjustments.items():
            if cargo_type in probs:
                probs[cargo_type] = adjustment
        
        # Normalize probabilities
        total_prob = sum(probs.values())
        probs = {k: v/total_prob for k, v in probs.items()}
        
        # Select based on adjusted probabilities
        cargo_types = list(probs.keys())
        probabilities = list(probs.values())
        
        return np.random.choice(cargo_types, p=probabilities)
    
    def _get_special_requirements(self, cargo_type: str, complexity_factors: List[str]) -> List[str]:
        """Get special requirements for cargo"""
        requirements = []
        
        if cargo_type == 'reefer':
            requirements.extend(['power_connection', 'temperature_control'])
        elif cargo_type == 'hazardous':
            requirements.extend(['segregation', 'special_handling'])
        elif cargo_type == 'oversized':
            requirements.extend(['special_lashing', 'oversized_equipment'])
        
        if 'weight_distribution' in complexity_factors:
            requirements.append('weight_distribution_constraint')
        
        return requirements

# Initialize scenario generator
scenario_generator = ScenarioGenerator()
print("Scenario generator initialized")
print(f"Available routes: {len(scenario_generator.global_routes)}")
print(f"Cargo types: {list(scenario_generator.cargo_types.keys())}")
```

## 4. Performance Analysis

Let's analyze the performance of the Human-AI Partnership system across different dimensions of decision quality and efficiency.

```python
def analyze_partnership_performance(decisions: List[PartnershipDecision]) -> Dict[str, Any]:
    """Comprehensive performance analysis of the partnership system"""
    
    # Decision level distribution
    decision_levels = defaultdict(int)
    for decision in decisions:
        decision_levels[decision.decision_level.value] += 1
    
    # Confidence analysis
    ai_confidences = [d.ai_confidence for d in decisions]
    human_confidences = [d.human_confidence for d in decisions]
    final_confidences = [d.final_confidence for d in decisions]
    
    # Decision time analysis
    decision_times = [d.decision_time for d in decisions]
    
    # Consensus analysis
    consensus_scores = [d.consensus_score for d in decisions]
    
    # Performance by complexity
    complexity_performance = analyze_by_complexity(decisions)
    
    return {
        'decision_levels': dict(decision_levels),
        'confidence_analysis': {
            'ai_mean': np.mean(ai_confidences),
            'ai_std': np.std(ai_confidences),
            'human_mean': np.mean(human_confidences),
            'human_std': np.std(human_confidences),
            'final_mean': np.mean(final_confidences),
            'final_std': np.std(final_confidences)
        },
        'time_analysis': {
            'mean_time': np.mean(decision_times),
            'median_time': np.median(decision_times),
            'std_time': np.std(decision_times),
            'min_time': np.min(decision_times),
            'max_time': np.max(decision_times)
        },
        'consensus_analysis': {
            'mean_consensus': np.mean(consensus_scores),
            'consensus_distribution': np.histogram(consensus_scores, bins=5, range=(0,1))[0]
        },
        'complexity_performance': complexity_performance
    }

def analyze_by_complexity(decisions: List[PartnershipDecision]) -> Dict[str, Any]:
    """Analyze performance by scenario complexity"""
    
    # Get scenarios from decision history
    scenarios = []
    for decision in decisions:
        # Find corresponding scenario (simplified)
        complexity = np.random.uniform(0.3, 0.9)  # Placeholder
        scenarios.append({
            'complexity': complexity,
            'decision_level': decision.decision_level.value,
            'final_confidence': decision.final_confidence,
            'decision_time': decision.decision_time
        })
    
    # Group by complexity levels
    low_complexity = [s for s in scenarios if s['complexity'] < 0.4]
    medium_complexity = [s for s in scenarios if 0.4 <= s['complexity'] < 0.7]
    high_complexity = [s for s in scenarios if s['complexity'] >= 0.7]
    
    return {
        'low_complexity': {
            'count': len(low_complexity),
            'avg_confidence': np.mean([s['final_confidence'] for s in low_complexity]) if low_complexity else 0,
            'avg_time': np.mean([s['decision_time'] for s in low_complexity]) if low_complexity else 0
        },
        'medium_complexity': {
            'count': len(medium_complexity),
            'avg_confidence': np.mean([s['final_confidence'] for s in medium_complexity]) if medium_complexity else 0,
            'avg_time': np.mean([s['decision_time'] for s in medium_complexity]) if medium_complexity else 0
        },
        'high_complexity': {
            'count': len(high_complexity),
            'avg_confidence': np.mean([s['final_confidence'] for s in high_complexity]) if high_complexity else 0,
            'avg_time': np.mean([s['decision_time'] for s in high_complexity]) if high_complexity else 0
        }
    }

# Analyze performance
performance_analysis = analyze_partnership_performance(partnership_decisions)

print("=== HUMAN-AI PARTNERSHIP PERFORMANCE ANALYSIS ===")
print("\nDecision Level Distribution:")
for level, count in performance_analysis['decision_levels'].items():
    percentage = (count / len(partnership_decisions)) * 100
    print(f"  {level}: {count} ({percentage:.1f}%)")

print("\nConfidence Analysis:")
conf_analysis = performance_analysis['confidence_analysis']
print(f"  AI Confidence: {conf_analysis['ai_mean']:.3f} ± {conf_analysis['ai_std']:.3f}")
print(f"  Human Confidence: {conf_analysis['human_mean']:.3f} ± {conf_analysis['human_std']:.3f}")
print(f"  Final Confidence: {conf_analysis['final_mean']:.3f} ± {conf_analysis['final_std']:.3f}")

print("\nDecision Time Analysis:")
time_analysis = performance_analysis['time_analysis']
print(f"  Mean Time: {time_analysis['mean_time']:.1f} minutes")
print(f"  Median Time: {time_analysis['median_time']:.1f} minutes")
print(f"  Range: {time_analysis['min_time']:.1f} - {time_analysis['max_time']:.1f} minutes")

print("\nConsensus Analysis:")
consensus_analysis = performance_analysis['consensus_analysis']
print(f"  Mean Consensus Score: {consensus_analysis['mean_consensus']:.3f}")

print("\nPerformance by Complexity:")
complexity_perf = performance_analysis['complexity_performance']
for complexity_level, metrics in complexity_perf.items():
    if metrics['count'] > 0:
        print(f"  {complexity_level}: {metrics['count']} scenarios, "
              f"avg confidence: {metrics['avg_confidence']:.3f}, "
              f"avg time: {metrics['avg_time']:.1f} min")
```

## 5. Comparison with Alternative Approaches

Let's compare the Human-AI Partnership approach with pure AI automation and traditional human-only approaches.

```python
def compare_approaches(decisions: List[PartnershipDecision]) -> Dict[str, Dict]:
    """Compare partnership approach with alternatives"""
    
    # Simulate alternative approaches
    pure_ai_decisions = simulate_pure_ai_approach(len(decisions))
    human_only_decisions = simulate_human_only_approach(len(decisions))
    
    # Calculate metrics for each approach
    partnership_metrics = calculate_approach_metrics(decisions, "Partnership")
    pure_ai_metrics = calculate_approach_metrics(pure_ai_decisions, "Pure AI")
    human_only_metrics = calculate_approach_metrics(human_only_decisions, "Human Only")
    
    return {
        'partnership': partnership_metrics,
        'pure_ai': pure_ai_metrics,
        'human_only': human_only_metrics
    }

def simulate_pure_ai_approach(num_scenarios: int) -> List[Dict]:
    """Simulate pure AI automation approach"""
    decisions = []
    
    for i in range(num_scenarios):
        complexity = np.random.uniform(0.3, 0.9)
        
        # AI confidence decreases with complexity
        base_confidence = 0.87
        complexity_penalty = complexity * 0.3
        confidence = max(0.4, base_confidence - complexity_penalty)
        
        # Fast but potentially less accurate for complex scenarios
        decision_time = 0.5 + complexity * 2.0  # 0.5 to 2.5 minutes
        
        # Accuracy drops with complexity
        accuracy = min(0.95, 0.95 - complexity * 0.2)
        
        decisions.append({
            'confidence': confidence,
            'decision_time': decision_time,
            'accuracy': accuracy,
            'complexity': complexity
        })
    
    return decisions

def simulate_human_only_approach(num_scenarios: int) -> List[Dict]:
    """Simulate traditional human-only approach"""
    decisions = []
    
    for i in range(num_scenarios):
        complexity = np.random.uniform(0.3, 0.9)
        
        # Human confidence can be high but variable
        base_confidence = 0.75
        experience_bonus = 0.15  # Experience helps with complexity
        variability = np.random.normal(0, 0.1)
        confidence = np.clip(base_confidence + experience_bonus + variability, 0.4, 0.95)
        
        # Slower but more thoughtful for complex scenarios
        decision_time = 8.0 + complexity * 12.0  # 8 to 20 minutes
        
        # Humans can handle novel situations better
        accuracy = min(0.92, 0.82 + complexity * 0.1)  # Actually improves with complexity up to a point
        
        decisions.append({
            'confidence': confidence,
            'decision_time': decision_time,
            'accuracy': accuracy,
            'complexity': complexity
        })
    
    return decisions

def calculate_approach_metrics(decisions: List[Dict], approach_name: str) -> Dict[str, float]:
    """Calculate performance metrics for an approach"""
    
    confidences = [d['confidence'] for d in decisions]
    decision_times = [d['decision_time'] for d in decisions]
    accuracies = [d['accuracy'] for d in decisions]
    complexities = [d['complexity'] for d in decisions]
    
    return {
        'avg_confidence': np.mean(confidences),
        'avg_decision_time': np.mean(decision_times),
        'avg_accuracy': np.mean(accuracies),
        'confidence_std': np.std(confidences),
        'time_std': np.std(decision_times),
        'accuracy_std': np.std(accuracies),
        'complexity_correlation': np.corrcoef(complexities, accuracies)[0, 1] if len(decisions) > 1 else 0
    }

# Compare approaches
approach_comparison = compare_approaches(partnership_decisions)

print("=== APPROACH COMPARISON ANALYSIS ===")
print("\nPerformance Metrics by Approach:")
print(f"{'Approach':<15} {'Confidence':<12} {'Time (min)':<12} {'Accuracy':<12} {'Complexity Corr':<15}")
print("-" * 70)

for approach, metrics in approach_comparison.items():
    print(f"{approach:<15} {metrics['avg_confidence']:<12.3f} {metrics['avg_decision_time']:<12.1f} "
          f"{metrics['avg_accuracy']:<12.3f} {metrics['complexity_correlation']:<15.3f}")

# Calculate improvements
partnership_metrics = approach_comparison['partnership']
pure_ai_metrics = approach_comparison['pure_ai']
human_only_metrics = approach_comparison['human_only']

print("\nPartnership Improvements:")
confidence_improvement_vs_ai = (partnership_metrics['avg_confidence'] - pure_ai_metrics['avg_confidence']) / pure_ai_metrics['avg_confidence'] * 100
accuracy_improvement_vs_ai = (partnership_metrics['avg_accuracy'] - pure_ai_metrics['avg_accuracy']) / pure_ai_metrics['avg_accuracy'] * 100
time_improvement_vs_human = (human_only_metrics['avg_decision_time'] - partnership_metrics['avg_decision_time']) / human_only_metrics['avg_decision_time'] * 100

print(f"  vs Pure AI: +{confidence_improvement_vs_ai:.1f}% confidence, +{accuracy_improvement_vs_ai:.1f}% accuracy")
print(f"  vs Human Only: -{time_improvement_vs_human:.1f}% decision time")

# Best approach by metric
best_confidence = max(approach_comparison.items(), key=lambda x: x[1]['avg_confidence'])
best_accuracy = max(approach_comparison.items(), key=lambda x: x[1]['avg_accuracy'])
best_speed = min(approach_comparison.items(), key=lambda x: x[1]['avg_decision_time'])

print(f"\nBest by Metric:")
print(f"  Confidence: {best_confidence[0]} ({best_confidence[1]['avg_confidence']:.3f})")
print(f"  Accuracy: {best_accuracy[0]} ({best_accuracy[1]['avg_accuracy']:.3f})")
print(f"  Speed: {best_speed[0]} ({best_speed[1]['avg_decision_time']:.1f} min)")
```

```python
class LearningSystem:
    """Learning and adaptation system for human-AI partnership"""
    
    def __init__(self):
        self.decision_history = []
        self.feedback_history = []
        self.performance_trends = defaultdict(list)
        self.adaptation_parameters = {
            'confidence_calibration_rate': 0.1,
            'decision_level_adjustment_rate': 0.05,
            'expert_selection_learning_rate': 0.15
        }
    
    def record_decision_outcome(self, decision: PartnershipDecision, 
                               actual_outcome: Dict[str, float]):
        """Record the actual outcome of a decision for learning"""
        
        outcome_record = {
            'decision_id': decision.scenario_id,
            'predicted_confidence': decision.final_confidence,
            'actual_performance': actual_outcome.get('performance_score', 0.8),
            'decision_level': decision.decision_level.value,
            'human_confidence': decision.human_confidence,
            'ai_confidence': decision.ai_confidence,
            'decision_time': decision.decision_time,
            'timestamp': datetime.now()
        }
        
        self.decision_history.append(outcome_record)
        
        # Update performance trends
        self.performance_trends['overall'].append(actual_outcome.get('performance_score', 0.8))
        self.performance_trends[decision.decision_level.value].append(actual_outcome.get('performance_score', 0.8))
        
        # Adapt system parameters
        self._adapt_parameters(outcome_record)
    
    def _adapt_parameters(self, outcome_record: Dict):
        """Adapt system parameters based on decision outcomes"""
        
        predicted = outcome_record['predicted_confidence']
        actual = outcome_record['actual_performance']
        
        # Confidence calibration adaptation
        confidence_error = predicted - actual
        calibration_adjustment = confidence_error * self.adaptation_parameters['confidence_calibration_rate']
        
        # Decision level adaptation
        if abs(confidence_error) > 0.2:  # Large error
            # Adjust decision level thresholds
            pass  # Implementation would adjust decision level selection logic
    
    def analyze_learning_progress(self) -> Dict[str, Any]:
        """Analyze learning progress over time"""
        
        if len(self.decision_history) < 10:
            return {'message': 'Insufficient data for learning analysis'}
        
        # Calculate learning curves
        overall_performance = self.performance_trends['overall']
        
        # Moving averages
        window_sizes = [5, 10, 15]
        learning_curves = {}
        
        for window_size in window_sizes:
            if len(overall_performance) >= window_size:
                moving_avg = []
                for i in range(window_size, len(overall_performance) + 1):
                    window_data = overall_performance[i-window_size:i]
                    moving_avg.append(np.mean(window_data))
                learning_curves[f'window_{window_size}'] = moving_avg
        
        # Calculate improvement rate
        if len(overall_performance) >= 20:
            early_performance = np.mean(overall_performance[:10])
            recent_performance = np.mean(overall_performance[-10:])
            improvement_rate = (recent_performance - early_performance) / early_performance * 100
        else:
            improvement_rate = 0
        
        return {
            'total_decisions': len(self.decision_history),
            'learning_curves': learning_curves,
            'improvement_rate': improvement_rate,
            'current_performance': overall_performance[-1] if overall_performance else 0,
            'performance_by_level': {
                level: np.mean(performance_list) if performance_list else 0
                for level, performance_list in self.performance_trends.items()
                if level != 'overall'
            }
        }

# Simulate learning with feedback
learning_system = LearningSystem()

print("=== LEARNING SYSTEM SIMULATION ===")
print("Simulating decision outcomes and learning adaptation...\n")

# Simulate outcomes for partnership decisions
for i, decision in enumerate(partnership_decisions):
    # Simulate actual outcome (with some realistic variation)
    base_performance = decision.final_confidence
    noise = np.random.normal(0, 0.1)
    actual_performance = np.clip(base_performance + noise, 0.3, 1.0)
    
    outcome = {
        'performance_score': actual_performance,
        'overstowage_reduction': np.random.uniform(15, 35),
        'handling_efficiency': np.random.uniform(0.8, 0.95),
        'safety_compliance': np.random.uniform(0.9, 1.0)
    }
    
    learning_system.record_decision_outcome(decision, outcome)
    
    if i < 5:  # Show first few
        print(f"Decision {i+1}: Predicted {decision.final_confidence:.3f}, Actual {actual_performance:.3f}")

# Analyze learning progress
learning_analysis = learning_system.analyze_learning_progress()

print(f"\n=== LEARNING ANALYSIS ===")
if 'message' in learning_analysis:
    print(learning_analysis['message'])
else:
    print(f"Total Decisions: {learning_analysis['total_decisions']}")
    print(f"Improvement Rate: {learning_analysis['improvement_rate']:.1f}%")
    print(f"Current Performance: {learning_analysis['current_performance']:.3f}")
    
    print("\nPerformance by Decision Level:")
    for level, performance in learning_analysis['performance_by_level'].items():
        print(f"  {level}: {performance:.3f}")
```
