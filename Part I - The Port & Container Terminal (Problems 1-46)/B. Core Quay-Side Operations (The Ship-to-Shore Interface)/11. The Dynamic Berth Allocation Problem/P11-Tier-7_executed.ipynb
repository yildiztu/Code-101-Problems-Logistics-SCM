# 11. The Dynamic Berth Allocation Problem
## Tier 7 — Human-AI Symbiotic Partnership (The Centaur Model)

### Goal
Learn how to design and implement a collaborative Human-AI partnership that combines human expertise with AI computational power to achieve superior berth allocation decisions.

### Key assumptions
- Human operators provide contextual knowledge and strategic oversight
- AI systems deliver computational optimization and pattern recognition
- Explainable AI ensures transparency in automated recommendations
- Decision modes adapt based on situation complexity and uncertainty
- Trust develops through consistent performance and transparent reasoning

### Approach (step-by-step)
1. **Design augmented intelligence architecture** with clear human-AI roles
2. **Implement explainable AI components** for transparent decision rationale
3. **Create collaborative decision framework** with AI-led, human-led, and collaborative modes
4. **Develop confidence scoring and trust metrics** for decision quality assessment
5. **Build visualization dashboard** for human-AI interaction and monitoring
6. **Simulate partnership performance** with real-world vessel allocation scenarios

### What to look for in the results
- Decision mode distribution (AI-led vs human-led vs collaborative)
- Performance improvements over pure AI or human-only approaches
- Trust scores and operator satisfaction metrics
- Explainable AI outputs with feature importance and confidence intervals
- Learning and adaptation patterns over time

### Concrete example (from the source)
During the Port Metropolis annual efficiency evaluation, the Human-AI partnership handled 1,247 vessel assignments over 6 months:
- AI-led mode: 68% of decisions (847 vessels) with 12 seconds average and 94% acceptance
- Collaborative mode: 24% of decisions (299 vessels) with 3.2 minutes average and 15% better solutions
- Human-led mode: 8% of decisions (101 vessels) with 8.4 minutes average and zero customer complaints
- Overall: 18% reduction vs pure AI, 31% reduction vs human-only, 97% operator satisfaction

```python
# Define data structures for Human-AI Partnership
class DecisionMode(Enum):
    """Decision modes for Human-AI collaboration"""
    AI_LED = "AI-Led"          # AI makes decisions with high confidence
    HUMAN_LED = "Human-Led"    # Human makes final decision
    COLLABORATIVE = "Collaborative"  # Human and AI work together

@dataclass
class Vessel:
    """Represents a vessel with enhanced characteristics for AI processing"""
    id: int
    arrival_time: int
    desired_departure: int
    length: int
    teu_capacity: int
    processing_times: Dict[int, int]
    priority: str = "Normal"  # VIP, High, Normal, Low
    customer_importance: float = 1.0  # 1.0 (normal) to 5.0 (VIP)
    schedule_flexibility: float = 1.0  # 0.0 (rigid) to 1.0 (flexible)
    historical_performance: float = 0.95  # Historical on-time performance

@dataclass
class Berth:
    """Represents a berth with detailed characteristics"""
    id: int
    capacity: int
    crane_count: int
    productivity: float  # TEU per hour
    location_preference: Dict[str, float] = field(default_factory=dict)  # Vessel type preferences
    current_utilization: float = 0.0

@dataclass
class AIDecision:
    """AI-generated decision with explainable components"""
    vessel_id: int
    recommended_berth: int
    recommended_start_time: int
    confidence_score: float  # 0.0 to 1.0
    feature_importance: Dict[str, float]
    alternative_solutions: List[Dict]
    reasoning: str
    processing_time_ms: int

@dataclass
class HumanDecision:
    """Human operator decision with context"""
    vessel_id: int
    final_berth: int
    final_start_time: int
    decision_mode: DecisionMode
    reasoning: str
    confidence_level: float  # Human's confidence in decision
    time_to_decide_seconds: int
    ai_agreement_score: float  # How much human agreed with AI recommendation

@dataclass
class PartnershipMetrics:
    """Metrics for evaluating Human-AI partnership performance"""
    total_decisions: int = 0
    ai_led_decisions: int = 0
    human_led_decisions: int = 0
    collaborative_decisions: int = 0
    ai_acceptance_rate: float = 0.0
    average_decision_time_seconds: float = 0.0
    trust_score: float = 0.0
    operator_satisfaction: float = 0.0
    performance_improvement_vs_ai: float = 0.0
    performance_improvement_vs_human: float = 0.0

print("Data structures defined for Human-AI Symbiotic Partnership")
```

```python
# Implement Explainable AI System
class ExplainableAISystem:
    """AI system with explainable decision capabilities"""
    
    def __init__(self, vessels: List[Vessel], berths: List[Berth]):
        self.vessels = vessels
        self.berths = berths
        self.feature_weights = {
            'vessel_priority': 0.25,
            'customer_importance': 0.20,
            'berth_productivity': 0.15,
            'schedule_flexibility': 0.10,
            'historical_performance': 0.10,
            'berth_preference_match': 0.10,
            'processing_time_efficiency': 0.10
        }
    
    def calculate_feature_importance(self, vessel: Vessel, berth: Berth, start_time: int) -> Dict[str, float]:
        """Calculate feature importance for a specific vessel-berth-time assignment"""
        
        features = {}
        
        # Vessel priority impact
        priority_score = {'VIP': 1.0, 'High': 0.75, 'Normal': 0.5, 'Low': 0.25}[vessel.priority]
        features['vessel_priority'] = priority_score * self.feature_weights['vessel_priority']
        
        # Customer importance
        features['customer_importance'] = (vessel.customer_importance / 5.0) * self.feature_weights['customer_importance']
        
        # Berth productivity
        features['berth_productivity'] = (berth.productivity / 200.0) * self.feature_weights['berth_productivity']
        
        # Schedule flexibility
        features['schedule_flexibility'] = vessel.schedule_flexibility * self.feature_weights['schedule_flexibility']
        
        # Historical performance
        features['historical_performance'] = vessel.historical_performance * self.feature_weights['historical_performance']
        
        # Berth preference match
        preference_score = berth.location_preference.get(vessel.priority, 0.5)
        features['berth_preference_match'] = preference_score * self.feature_weights['berth_preference_match']
        
        # Processing time efficiency
        processing_time = vessel.processing_times[berth.id]
        efficiency_score = max(0, 1.0 - processing_time / 10.0)  # Normalize by max expected time
        features['processing_time_efficiency'] = efficiency_score * self.feature_weights['processing_time_efficiency']
        
        return features
    
    def generate_decision(self, vessel: Vessel, current_schedule: List[Dict]) -> AIDecision:
        """Generate AI decision with explainable components"""
        
        start_time = time.time()
        
        # Evaluate all possible assignments
        candidates = []
        
        for berth in self.berths:
            # Check capacity constraint
            if vessel.length > berth.capacity:
                continue
            
            # Find earliest feasible start time
            earliest_start = vessel.arrival_time
            processing_time = vessel.processing_times[berth.id]
            
            # Check for conflicts with current schedule
            for scheduled in current_schedule:
                if (scheduled['berth_id'] == berth.id and 
                    not (earliest_start + processing_time <= scheduled['start_time'] or
                         earliest_start >= scheduled['completion_time'])):
                    earliest_start = scheduled['completion_time']
            
            # Calculate decision score
            features = self.calculate_feature_importance(vessel, berth, earliest_start)
            decision_score = sum(features.values())
            
            # Add penalty for waiting time
            waiting_time = max(0, earliest_start - vessel.arrival_time)
            waiting_penalty = waiting_time * 0.05
            decision_score -= waiting_penalty
            
            candidates.append({
                'berth_id': berth.id,
                'start_time': earliest_start,
                'completion_time': earliest_start + processing_time,
                'decision_score': decision_score,
                'features': features,
                'waiting_time': waiting_time,
                'processing_time': processing_time
            })
        
        # Sort candidates by decision score
        candidates.sort(key=lambda x: x['decision_score'], reverse=True)
        
        if not candidates:
            # No feasible assignment found
            return AIDecision(
                vessel_id=vessel.id,
                recommended_berth=-1,
                recommended_start_time=-1,
                confidence_score=0.0,
                feature_importance={},
                alternative_solutions=[],
                reasoning="No feasible berth assignment available",
                processing_time_ms=int((time.time() - start_time) * 1000)
            )
        
        # Select best candidate
        best_candidate = candidates[0]
        
        # Calculate confidence score based on decision score margin
        if len(candidates) > 1:
            score_margin = best_candidate['decision_score'] - candidates[1]['decision_score']
            confidence_score = min(1.0, max(0.3, score_margin * 2))
        else:
            confidence_score = 0.7
        
        # Generate alternative solutions
        alternatives = []
        for i, candidate in enumerate(candidates[1:4]):  # Top 3 alternatives
            alternatives.append({
                'berth_id': candidate['berth_id'],
                'start_time': candidate['start_time'],
                'decision_score': candidate['decision_score'],
                'trade_offs': f"Lower score by {best_candidate['decision_score'] - candidate['decision_score']:.2f}"
            })
        
        # Generate reasoning
        reasoning = f"Recommended Berth {best_candidate['berth_id']} at time {best_candidate['start_time']} "
        reasoning += f"based on high decision score ({best_candidate['decision_score']:.2f}). "
        reasoning += f"Key factors: vessel priority ({vessel.priority}), "
        reasoning += f"berth productivity ({berth.productivity} TEU/h), "
        reasoning += f"waiting time ({best_candidate['waiting_time']:.1f}h)."
        
        return AIDecision(
            vessel_id=vessel.id,
            recommended_berth=best_candidate['berth_id'],
            recommended_start_time=best_candidate['start_time'],
            confidence_score=confidence_score,
            feature_importance=best_candidate['features'],
            alternative_solutions=alternatives,
            reasoning=reasoning,
            processing_time_ms=int((time.time() - start_time) * 1000)
        )

print("Explainable AI System implemented")
```

```python
# Implement Human Operator System
class HumanOperator:
    """Human operator with expertise and decision-making capabilities"""
    
    def __init__(self, experience_years: int = 10, specialization: str = "Berth Allocation"):
        self.experience_years = experience_years
        self.specialization = specialization
        self.trust_score = 0.7  # Initial trust in AI system
        self.decision_history = []
        self.expertise_patterns = {
            'VIP_vessels': 0.95,  # Expertise in handling VIP vessels
            'emergency_situations': 0.88,
            'capacity_constraints': 0.82,
            'schedule_optimization': 0.90,
            'customer_relations': 0.85
        }
    
    def assess_situation_complexity(self, vessel: Vessel, ai_decision: AIDecision) -> float:
        """Assess the complexity of the current situation (0.0 to 1.0)"""
        
        complexity_factors = {
            'vip_vessel': 1.0 if vessel.priority == 'VIP' else 0.0,
            'high_importance': (vessel.customer_importance - 2.0) / 3.0 if vessel.customer_importance > 2.0 else 0.0,
            'low_confidence': max(0, (0.7 - ai_decision.confidence_score) / 0.7),
            'tight_schedule': max(0, (vessel.desired_departure - vessel.arrival_time - 8) / 8),
            'low_flexibility': max(0, (1.0 - vessel.schedule_flexibility)),
            'performance_issues': max(0, (0.9 - vessel.historical_performance) / 0.9)
        }
        
        # Weighted complexity score
        weights = {'vip_vessel': 0.3, 'high_importance': 0.2, 'low_confidence': 0.2, 
                  'tight_schedule': 0.1, 'low_flexibility': 0.1, 'performance_issues': 0.1}
        
        complexity = sum(complexity_factors[factor] * weights[factor] for factor in complexity_factors)
        return min(1.0, complexity)
    
    def determine_decision_mode(self, vessel: Vessel, ai_decision: AIDecision) -> DecisionMode:
        """Determine the appropriate decision mode based on situation assessment"""
        
        complexity = self.assess_situation_complexity(vessel, ai_decision)
        
        # Decision logic based on complexity and AI confidence
        if vessel.priority == 'VIP' or complexity > 0.7:
            return DecisionMode.HUMAN_LED
        elif ai_decision.confidence_score > 0.85 and complexity < 0.3:
            return DecisionMode.AI_LED
        else:
            return DecisionMode.COLLABORATIVE
    
    def make_human_decision(self, vessel: Vessel, ai_decision: AIDecision, 
                           current_schedule: List[Dict]) -> HumanDecision:
        """Make human decision with AI input consideration"""
        
        start_time = time.time()
        decision_mode = self.determine_decision_mode(vessel, ai_decision)
        
        # Simulate human decision-making process
        if decision_mode == DecisionMode.AI_LED:
            # Human accepts AI recommendation with minor review
            final_berth = ai_decision.recommended_berth
            final_start_time = ai_decision.recommended_start_time
            reasoning = f"Accepted AI recommendation: {ai_decision.reasoning}"
            confidence_level = ai_decision.confidence_score * 0.9  # Slightly conservative
            ai_agreement_score = 1.0
            
        elif decision_mode == DecisionMode.HUMAN_LED:
            # Human makes independent decision
            # Simulate human expertise-based decision
            if vessel.priority == 'VIP':
                # Prefer best berth for VIP vessels
                final_berth = 1  # Best berth
                final_start_time = vessel.arrival_time  # No waiting for VIP
            else:
                # Use human heuristics
                final_berth = (vessel.id % len(berths)) + 1
                final_start_time = max(vessel.arrival_time, 
                                      (vessel.id * 2) % 24)  # Human heuristic
            
            reasoning = f"Human expertise applied for {vessel.priority} vessel. "
            reasoning += f"Selected Berth {final_berth} based on experience patterns."
            confidence_level = 0.85  # High confidence in own expertise
            ai_agreement_score = 0.3 if final_berth != ai_decision.recommended_berth else 0.8
            
        else:  # COLLABORATIVE
            # Human collaborates with AI, may modify recommendation
            if ai_decision.confidence_score > 0.7 and vessel.priority != 'VIP':
                # Accept AI recommendation with minor adjustments
                final_berth = ai_decision.recommended_berth
                final_start_time = ai_decision.recommended_start_time
                if vessel.schedule_flexibility > 0.7 and ai_decision.recommended_start_time > vessel.arrival_time:
                    # Human suggests slight delay for better overall flow
                    final_start_time += 1
                reasoning = f"Collaborative: AI recommendation refined with human insight. "
                reasoning += f"Adjusted start time for optimal terminal flow."
                ai_agreement_score = 0.8
            else:
                # Human considers alternatives
                if ai_decision.alternative_solutions:
                    alt = ai_decision.alternative_solutions[0]
                    final_berth = alt['berth_id']
                    final_start_time = alt['start_time']
                    reasoning = f"Collaborative: Selected AI alternative {alt['trade_offs']}. "
                    reasoning += f"Human judgment favored this option."
                else:
                    final_berth = ai_decision.recommended_berth
                    final_start_time = ai_decision.recommended_start_time
                    reasoning = "Collaborative: AI recommendation accepted after review."
                ai_agreement_score = 0.6
            
            confidence_level = (ai_decision.confidence_score + 0.85) / 2  # Balanced confidence
        
        decision_time = int(time.time() - start_time)
        
        return HumanDecision(
            vessel_id=vessel.id,
            final_berth=final_berth,
            final_start_time=final_start_time,
            decision_mode=decision_mode,
            reasoning=reasoning,
            confidence_level=confidence_level,
            time_to_decide_seconds=decision_time,
            ai_agreement_score=ai_agreement_score
        )
    
    def update_trust_score(self, decision: HumanDecision, actual_performance: float):
        """Update trust score based on decision performance"""
        if actual_performance > 0.8:  # Good performance
            if decision.decision_mode == DecisionMode.AI_LED:
                self.trust_score = min(1.0, self.trust_score + 0.02)
            elif decision.ai_agreement_score > 0.7:
                self.trust_score = min(1.0, self.trust_score + 0.01)
        elif actual_performance < 0.6:  # Poor performance
            if decision.decision_mode == DecisionMode.AI_LED:
                self.trust_score = max(0.3, self.trust_score - 0.03)
        
        self.decision_history.append(decision)

print("Human Operator System implemented")
```

```python
# Implement Human-AI Partnership Manager
class HumanAIPartnershipManager:
    """Manages the collaboration between human operators and AI systems"""
    
    def __init__(self, vessels: List[Vessel], berths: List[Berth]):
        self.vessels = vessels
        self.berths = berths
        self.ai_system = ExplainableAISystem(vessels, berths)
        self.human_operator = HumanOperator()
        self.current_schedule = []
        self.metrics = PartnershipMetrics()
        self.decision_log = []
    
    def process_vessel_arrival(self, vessel: Vessel) -> HumanDecision:
        """Process a single vessel arrival through Human-AI partnership"""
        
        # Generate AI decision
        ai_decision = self.ai_system.generate_decision(vessel, self.current_schedule)
        
        # Human makes final decision
        human_decision = self.human_operator.make_human_decision(
            vessel, ai_decision, self.current_schedule
        )
        
        # Update schedule
        processing_time = vessel.processing_times[human_decision.final_berth]
        schedule_entry = {
            'vessel_id': vessel.id,
            'berth_id': human_decision.final_berth,
            'start_time': human_decision.final_start_time,
            'completion_time': human_decision.final_start_time + processing_time,
            'processing_time': processing_time,
            'decision_mode': human_decision.decision_mode.value,
            'ai_confidence': ai_decision.confidence_score,
            'human_confidence': human_decision.confidence_level
        }
        
        self.current_schedule.append(schedule_entry)
        self.decision_log.append({
            'vessel': vessel,
            'ai_decision': ai_decision,
            'human_decision': human_decision,
            'schedule_entry': schedule_entry
        })
        
        # Update metrics
        self.update_metrics(human_decision)
        
        return human_decision
    
    def update_metrics(self, decision: HumanDecision):
        """Update partnership metrics"""
        self.metrics.total_decisions += 1
        
        if decision.decision_mode == DecisionMode.AI_LED:
            self.metrics.ai_led_decisions += 1
        elif decision.decision_mode == DecisionMode.HUMAN_LED:
            self.metrics.human_led_decisions += 1
        else:
            self.metrics.collaborative_decisions += 1
        
        # Update AI acceptance rate
        if decision.decision_mode == DecisionMode.AI_LED:
            acceptance_weight = 1.0
        elif decision.ai_agreement_score > 0.7:
            acceptance_weight = 0.8
        else:
            acceptance_weight = 0.3
        
        self.metrics.ai_acceptance_rate = (
            (self.metrics.ai_acceptance_rate * (self.metrics.total_decisions - 1) + acceptance_weight) /
            self.metrics.total_decisions
        )
        
        # Update average decision time
        self.metrics.average_decision_time_seconds = (
            (self.metrics.average_decision_time_seconds * (self.metrics.total_decisions - 1) + 
             decision.time_to_decide_seconds) / self.metrics.total_decisions
        )
        
        # Update trust score
        self.metrics.trust_score = self.human_operator.trust_score
    
    def simulate_partnership_operation(self) -> Dict:
        """Simulate the complete Human-AI partnership operation"""
        
        print("Simulating Human-AI Partnership Operation...")
        print("=" * 50)
        
        # Process all vessels
        for vessel in self.vessels:
            print(f"\nProcessing Vessel {vessel.id} ({vessel.priority} priority)...")
            decision = self.process_vessel_arrival(vessel)
            
            print(f"  Decision Mode: {decision.decision_mode.value}")
            print(f"  Final Assignment: Berth {decision.final_berth} at time {decision.final_start_time}")
            print(f"  Decision Time: {decision.time_to_decide_seconds} seconds")
            print(f"  AI Agreement: {decision.ai_agreement_score:.2f}")
            print(f"  Reasoning: {decision.reasoning}")
        
        # Calculate performance metrics
        self.calculate_performance_metrics()
        
        return {
            'schedule': self.current_schedule,
            'metrics': self.metrics,
            'decision_log': self.decision_log
        }
    
    def calculate_performance_metrics(self):
        """Calculate performance metrics for the partnership"""
        
        # Simulate performance comparison with pure AI and pure human
        total_service_time = sum(
            entry['completion_time'] - entry['start_time'] + 
            max(0, entry['start_time'] - self.vessels[entry['vessel_id']-1].arrival_time)
            for entry in self.current_schedule
        )
        
        # Simulated baselines (for demonstration)
        pure_ai_service_time = total_service_time * 1.18  # 18% worse
        pure_human_service_time = total_service_time * 1.31  # 31% worse
        
        self.metrics.performance_improvement_vs_ai = (
            (pure_ai_service_time - total_service_time) / pure_ai_service_time * 100
        )
        
        self.metrics.performance_improvement_vs_human = (
            (pure_human_service_time - total_service_time) / pure_human_service_time * 100
        )
        
        # Calculate operator satisfaction (based on trust and decision quality)
        self.metrics.operator_satisfaction = (
            self.metrics.trust_score * 0.6 + 
            self.metrics.ai_acceptance_rate * 0.4
        )

print("Human-AI Partnership Manager implemented")
```
