# 15. The Vessel Re-stow Problem (Transshipment)
## Tier 7: The Human-AI Symbiotic Partnership (The Centaur Model)

### Key assumptions
- Human experts and AI systems collaborate as symbiotic partners
- AI provides transparent reasoning and explainable recommendations
- Humans contribute domain expertise, contextual understanding, and ethical judgment
- Trust develops through successful collaborations and learning from each other

### Approach (step-by-step)
1. **Partnership framework**: Define roles and responsibilities for human-AI collaboration
2. **Explainable AI**: Develop interpretable models with transparent decision logic
3. **Trust building**: Implement confidence-based authority allocation
4. **Collaborative decision making**: Human-AI joint decision processes
5. **Learning loop**: Mutual learning and adaptation over time
6. **Performance tracking**: Measure partnership effectiveness and trust development
7. **Continuous improvement**: Refine collaboration based on experience

### What to look for in the results
- Trust evolution and partnership development over time
- Decision quality improvement through human-AI collaboration
- Explainable AI reasoning and transparency metrics
- Authority allocation patterns and confidence-based decisions
- Learning curves for both human and AI partners
- Comparative performance vs pure AI or pure human approaches

### Concrete example (from the source)
A human-AI partnership system deployed with 3 terminal operators and 2 AI agents for vessel re-stow optimization. Over 6 months of operation, the partnership achieved 94% agreement rate on critical decisions, with trust scores improving from 0.62 to 0.89. The collaborative approach reduced container moves by 42% compared to human-only decisions and 28% compared to AI-only decisions. Human experts contributed contextual knowledge that AI missed, while AI provided data-driven insights that humans overlooked.

- **Human oversight**: Ethical constraints vs purely objective optimization
- **Collaborative learning**: Mutual improvement vs one-sided learning

**Advantages over ACO (Tier 3):**
- **Domain expertise**: Human knowledge integration vs heuristic search
- **Contextual adaptation**: Human situational understanding vs fixed parameters

- **Intelligent collaboration**: AI augmentation vs hand-crafted rules

**Disadvantages:**

**When to use this Tier:**

```python
# Import required packages (all open-source)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass
from typing import List, Dict, Tuple, Optional, Union
import random
import time
from datetime import datetime, timedelta
import copy
import warnings
warnings.filterwarnings('ignore')

# Set up visualization style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
@dataclass
class HumanExpert:
    """Represents a human expert in the partnership"""
    id: str
    name: str
    expertise_area: str
    experience_years: int
    trust_score: float = 0.5  # Initial trust in AI
    confidence_level: float = 0.7
    availability: float = 0.9  # Probability of being available
    
@dataclass
class AIAgent:
    """Represents an AI agent in the partnership"""
    id: str
    name: str
    capability: str
    accuracy: float = 0.8
    explainability_score: float = 0.7
    confidence_threshold: float = 0.8
    learning_rate: float = 0.1
    
@dataclass
class DecisionContext:
    """Context for human-AI decision making"""
    problem_id: str
    timestamp: datetime
    complexity_level: float  # 0-1 scale
    criticality_level: float  # 0-1 scale
    uncertainty_level: float  # 0-1 scale
    time_pressure: float  # 0-1 scale
    
@dataclass
class AIDecision:
    """AI decision with explainable reasoning"""
    agent_id: str
    recommendation: Dict
    confidence: float
    reasoning: List[str]
    supporting_data: Dict
    uncertainty_factors: List[str]
    
@dataclass
class HumanDecision:
    """Human expert decision with domain knowledge"""
    expert_id: str
    decision: Dict
    confidence: float
    domain_insights: List[str]
    contextual_factors: List[str]
    ethical_considerations: List[str]
    
@dataclass
class CollaborativeDecision:
    """Final collaborative decision with authority allocation"""
    final_decision: Dict
    authority_distribution: Dict[str, float]  # Who had how much influence
    agreement_level: float  # 0-1 scale
    trust_update: Dict[str, float]  # Trust score changes
    learning_outcomes: List[str]
    decision_rationale: str
```

```python
class ExplainableAI:
    """Explainable AI system for vessel re-stow optimization"""
    
    def __init__(self, agent: AIAgent):
        self.agent = agent
        self.decision_history = []
        self.explanation_templates = {
            'container_move': "Moving container {container} from {source} to {target} reduces {metric} by {value}",
            'port_segregation': "Segregating {port} containers to bay {bay} improves {efficiency} by {percent}%",
            'weight_balance': "Rebalancing weight distribution improves vessel stability by {factor}",
            'accessibility': "Repositioning containers improves accessibility for {sequence} operations"
        }
    
    def analyze_re_stow_problem(self, containers: List, current_positions: List, 
                               target_positions: List) -> AIDecision:
        """Analyze vessel re-stow problem with explainable reasoning"""
        
        # Calculate optimization metrics
        total_moves = len([c for c in containers if c.current_position != c.target_position])
        move_cost = total_moves * 150.0
        
        # Generate AI recommendation
        recommendation = {
            'total_moves': total_moves,
            'total_cost': move_cost,
            'suggested_moves': self._generate_move_suggestions(containers),
            'optimization_strategy': 'cost_efficient_segregation'
        }
        
        # Generate explainable reasoning
        reasoning = []
        reasoning.append(f"Analysis of {len(containers)} containers identified {total_moves} required moves")
        reasoning.append(f"Total estimated cost: ${move_cost:.2f}")
        reasoning.append("Optimization prioritizes port segregation while minimizing moves")
        reasoning.append("Weight distribution constraints are satisfied in all proposed moves")
        
        # Identify uncertainty factors
        uncertainty_factors = []
        if total_moves > len(containers) * 0.5:
            uncertainty_factors.append("High number of required moves may affect operational efficiency")
        if move_cost > 1000:
            uncertainty_factors.append("High cost may require budget approval")
        
        # Calculate confidence based on problem complexity
        complexity = min(1.0, total_moves / len(containers))
        confidence = max(0.5, self.agent.accuracy * (1 - complexity * 0.3))
        
        return AIDecision(
            agent_id=self.agent.id,
            recommendation=recommendation,
            confidence=confidence,
            reasoning=reasoning,
            supporting_data={'containers_analyzed': len(containers), 'moves_required': total_moves},
            uncertainty_factors=uncertainty_factors
        )
    
    def _generate_move_suggestions(self, containers: List) -> List[Dict]:
        """Generate specific container move suggestions"""
        suggestions = []
        
        for container in containers[:5]:  # Top 5 containers
            if container.current_position != container.target_position:
                suggestions.append({
                    'container_id': container.id,
                    'from_position': container.current_position,
                    'to_position': container.target_position,
                    'reason': f"Port segregation: {container.destination_port}"
                })
        
        return suggestions
    
    def update_from_feedback(self, feedback: Dict):
        """Update AI agent based on human feedback"""
        if 'accuracy_feedback' in feedback:
            # Update accuracy based on feedback
            self.agent.accuracy += self.agent.learning_rate * (feedback['accuracy_feedback'] - self.agent.accuracy)
            self.agent.accuracy = max(0.5, min(0.95, self.agent.accuracy))
        
        if 'explainability_feedback' in feedback:
            self.agent.explainability_score += self.agent.learning_rate * \
                (feedback['explainability_feedback'] - self.agent.explainability_score)
            self.agent.explainability_score = max(0.5, min(0.95, self.agent.explainability_score))
```

```python
class HumanExpertSystem:
    """Human expert decision system with domain knowledge"""
    
    def __init__(self, expert: HumanExpert):
        self.expert = expert
        self.decision_history = []
        self.domain_knowledge = self._initialize_domain_knowledge()
    
    def _initialize_domain_knowledge(self) -> Dict:
        """Initialize domain knowledge based on expertise area"""
        knowledge = {
            'vessel_operations': {
                'weight_constraints': 'Heavy containers should be placed in lower tiers and central bays',
                'accessibility': 'High-priority discharge containers should be easily accessible',
                'stability': 'Weight distribution must maintain vessel stability during operations'
            },
            'port_operations': {
                'segregation_efficiency': 'Same-port containers should be grouped together',
                'handling_equipment': 'Consider crane reach and equipment limitations',
                'operational_flow': 'Minimize interference between different discharge ports'
            },
            'safety_compliance': {
                'weight_limits': 'Never exceed position weight limits',
                'stacking_rules': 'Follow proper stacking height restrictions',
                'hazardous_materials': 'Special handling for dangerous goods'
            }
        }
        
        return knowledge.get(self.expert.expertise_area, {})
    
    def analyze_re_stow_problem(self, containers: List, ai_decision: AIDecision, 
                               context: DecisionContext) -> HumanDecision:
        """Human expert analysis of AI recommendation"""
        
        # Evaluate AI recommendation
        ai_moves = ai_decision.recommendation['total_moves']
        ai_cost = ai_decision.recommendation['total_cost']
        
        # Apply domain knowledge
        domain_insights = []
        contextual_factors = []
        ethical_considerations = []
        
        # Domain insights based on expertise
        if self.expert.expertise_area == 'vessel_operations':
            domain_insights.append("Weight distribution should be prioritized for vessel stability")
            domain_insights.append("Consider tidal conditions and vessel trim")
        elif self.expert.expertise_area == 'port_operations':
            domain_insights.append("Terminal equipment availability affects move feasibility")
            domain_insights.append("Consider berth scheduling constraints")
        elif self.expert.expertise_area == 'safety_compliance':
            domain_insights.append("All moves must comply with safety regulations")
            domain_insights.append("Verify container weight limits and stacking rules")
        
        # Contextual factors
        if context.time_pressure > 0.7:
            contextual_factors.append("High time pressure - prioritize speed over optimization")
        if context.criticality_level > 0.8:
            contextual_factors.append("Critical operation - ensure safety and compliance")
        if context.uncertainty_level > 0.6:
            contextual_factors.append("High uncertainty - conservative approach recommended")
        
        # Ethical considerations
        ethical_considerations.append("Ensure fair treatment of all shipping lines")
        ethical_considerations.append("Minimize environmental impact of re-stow operations")
        ethical_considerations.append("Maintain transparency in decision making")
        
        # Make human decision
        human_decision = self._make_human_decision(ai_decision, context)
        
        # Calculate confidence based on experience and context
        experience_factor = min(1.0, self.expert.experience_years / 20)
        context_factor = 1 - (context.complexity_level * 0.3 + context.uncertainty_level * 0.2)
        confidence = self.expert.confidence_level * experience_factor * context_factor
        
        return HumanDecision(
            expert_id=self.expert.id,
            decision=human_decision,
            confidence=confidence,
            domain_insights=domain_insights,
            contextual_factors=contextual_factors,
            ethical_considerations=ethical_considerations
        )
    
    def _make_human_decision(self, ai_decision: AIDecision, context: DecisionContext) -> Dict:
        """Make human decision based on AI analysis and domain knowledge"""
        
        # Start with AI recommendation
        human_decision = ai_decision.recommendation.copy()
        
        # Apply human adjustments based on context
        if context.time_pressure > 0.7:
            # Reduce moves for time pressure
            human_decision['total_moves'] = max(1, human_decision['total_moves'] - 1)
            human_decision['total_cost'] = human_decision['total_moves'] * 150.0
            human_decision['optimization_strategy'] = 'time_priority'
        
        if context.criticality_level > 0.8:
            # Add safety margin for critical operations
            human_decision['safety_margin'] = 0.1
            human_decision['additional_checks'] = ['weight_verification', 'stability_analysis']
        
        return human_decision
    
    def update_trust(self, ai_performance: float):
        """Update trust in AI based on performance"""
        trust_change = 0.1 * (ai_performance - 0.7)  # Positive if AI performs well
        self.expert.trust_score += trust_change
        self.expert.trust_score = max(0.1, min(0.95, self.expert.trust_score))
```

```python
class CollaborativeDecisionEngine:
    """Human-AI collaborative decision making engine"""
    
    def __init__(self):
        self.ai_agents = []
        self.human_experts = []
        self.decision_history = []
        self.partnership_metrics = {
            'total_decisions': 0,
            'agreement_rate': 0.0,
            'average_trust': 0.0,
            'decision_quality': 0.0
        }
    
    def add_ai_agent(self, agent: AIAgent):
        """Add AI agent to partnership"""
        self.ai_agents.append(agent)
    
    def add_human_expert(self, expert: HumanExpert):
        """Add human expert to partnership"""
        self.human_experts.append(expert)
    
    def make_collaborative_decision(self, containers: List, context: DecisionContext) -> CollaborativeDecision:
        """Make collaborative human-AI decision"""
        
        # Get AI decisions
        ai_decisions = []
        for agent in self.ai_agents:
            explainable_ai = ExplainableAI(agent)
            ai_decision = explainable_ai.analyze_re_stow_problem(containers, [], [])
            ai_decisions.append(ai_decision)
        
        # Get human decisions
        human_decisions = []
        for expert in self.human_experts:
            if random.random() < expert.availability:  # Check availability
                human_system = HumanExpertSystem(expert)
                # Use first AI decision for human analysis
                human_decision = human_system.analyze_re_stow_problem(
                    containers, ai_decisions[0], context
                )
                human_decisions.append(human_decision)
        
        # Combine decisions using authority allocation
        final_decision, authority_dist, agreement = self._combine_decisions(
            ai_decisions, human_decisions, context
        )
        
        # Calculate trust updates
        trust_updates = self._calculate_trust_updates(ai_decisions, human_decisions, agreement)
        
        # Generate learning outcomes
        learning_outcomes = self._generate_learning_outcomes(ai_decisions, human_decisions, agreement)
        
        # Create decision rationale
        rationale = self._create_decision_rationale(ai_decisions, human_decisions, agreement)
        
        # Update partnership metrics
        self._update_partnership_metrics(agreement, trust_updates)
        
        return CollaborativeDecision(
            final_decision=final_decision,
            authority_distribution=authority_dist,
            agreement_level=agreement,
            trust_update=trust_updates,
            learning_outcomes=learning_outcomes,
            decision_rationale=rationale
        )
    
    def _combine_decisions(self, ai_decisions: List[AIDecision], 
                          human_decisions: List[HumanDecision], 
                          context: DecisionContext) -> Tuple[Dict, Dict, float]:
        """Combine AI and human decisions with authority allocation"""
        
        # Calculate authority weights based on confidence and trust
        authority_weights = {}
        
        # AI authority based on confidence and human trust
        for i, ai_decision in enumerate(ai_decisions):
            agent_id = ai_decision.agent_id
            avg_trust = np.mean([expert.trust_score for expert in self.human_experts])
            authority_weights[agent_id] = ai_decision.confidence * avg_trust
        
        # Human authority based on confidence and experience
        for human_decision in human_decisions:
            expert_id = human_decision.expert_id
            expert = next(e for e in self.human_experts if e.id == expert_id)
            experience_factor = min(1.0, expert.experience_years / 20)
            authority_weights[expert_id] = human_decision.confidence * experience_factor
        
        # Normalize authority weights
        total_authority = sum(authority_weights.values())
        if total_authority > 0:
            authority_weights = {k: v/total_authority for k, v in authority_weights.items()}
        
        # Combine decisions
        final_decision = {
            'total_moves': 0,
            'total_cost': 0.0,
            'optimization_strategy': 'collaborative'
        }
        
        # Weighted average of AI decisions
        for ai_decision in ai_decisions:
            weight = authority_weights[ai_decision.agent_id]
            final_decision['total_moves'] += weight * ai_decision.recommendation['total_moves']
            final_decision['total_cost'] += weight * ai_decision.recommendation['total_cost']
        
        # Adjust based on human decisions
        for human_decision in human_decisions:
            weight = authority_weights[human_decision.expert_id]
            final_decision['total_moves'] += weight * human_decision.decision.get('total_moves', 0)
            final_decision['total_cost'] += weight * human_decision.decision.get('total_cost', 0)
        
        # Round to integers
        final_decision['total_moves'] = int(round(final_decision['total_moves']))
        final_decision['total_cost'] = final_decision['total_moves'] * 150.0
        
        # Calculate agreement level
        if len(ai_decisions) > 0 and len(human_decisions) > 0:
            ai_avg_moves = np.mean([d.recommendation['total_moves'] for d in ai_decisions])
            human_avg_moves = np.mean([d.decision.get('total_moves', 0) for d in human_decisions])
            agreement = 1.0 - abs(ai_avg_moves - human_avg_moves) / max(ai_avg_moves, human_avg_moves, 1)
        else:
            agreement = 0.5  # Neutral if missing participants
        
        return final_decision, authority_weights, agreement
    
    def _calculate_trust_updates(self, ai_decisions: List[AIDecision], 
                                human_decisions: List[HumanDecision], 
                                agreement: float) -> Dict[str, float]:
        """Calculate trust updates based on decision agreement"""
        
        trust_updates = {}
        
        # Update human trust in AI based on agreement
        for expert in self.human_experts:
            trust_change = 0.05 * (agreement - 0.5)  # Positive if high agreement
            trust_updates[expert.id] = trust_change
            expert.trust_score += trust_change
            expert.trust_score = max(0.1, min(0.95, expert.trust_score))
        
        return trust_updates
    
    def _generate_learning_outcomes(self, ai_decisions: List[AIDecision], 
                                  human_decisions: List[HumanDecision], 
                                  agreement: float) -> List[str]:
        """Generate learning outcomes from collaboration"""
        
        outcomes = []
        
        if agreement > 0.8:
            outcomes.append("High agreement indicates strong alignment between AI and human expertise")
        elif agreement < 0.4:
            outcomes.append("Low agreement suggests need for better AI training or human-AI communication")
        
        if len(human_decisions) > 0:
            avg_confidence = np.mean([d.confidence for d in human_decisions])
            if avg_confidence > 0.8:
                outcomes.append("Human experts show high confidence in collaborative decisions")
        
        if len(ai_decisions) > 0:
            avg_ai_confidence = np.mean([d.confidence for d in ai_decisions])
            if avg_ai_confidence > 0.8:
                outcomes.append("AI agents demonstrate high confidence in recommendations")
        
        return outcomes
    
    def _create_decision_rationale(self, ai_decisions: List[AIDecision], 
                                  human_decisions: List[HumanDecision], 
                                  agreement: float) -> str:
        """Create rationale for final decision"""
        
        rationale_parts = []
        
        rationale_parts.append(f"Collaborative decision made with {agreement:.1%} agreement level")
        
        if len(ai_decisions) > 0:
            ai_reasoning = ai_decisions[0].reasoning[0] if ai_decisions[0].reasoning else "AI analysis completed"
            rationale_parts.append(f"AI reasoning: {ai_reasoning}")
        
        if len(human_decisions) > 0:
            human_insight = human_decisions[0].domain_insights[0] if human_decisions[0].domain_insights else "Human expertise applied"
            rationale_parts.append(f"Human insight: {human_insight}")
        
        return " | ".join(rationale_parts)
    
    def _update_partnership_metrics(self, agreement: float, trust_updates: Dict[str, float]):
        """Update overall partnership metrics"""
        
        self.partnership_metrics['total_decisions'] += 1
        
        # Update agreement rate (moving average)
        current_agreement = self.partnership_metrics['agreement_rate']
        n = self.partnership_metrics['total_decisions']
        self.partnership_metrics['agreement_rate'] = (current_agreement * (n-1) + agreement) / n
        
        # Update average trust
        if self.human_experts:
            avg_trust = np.mean([expert.trust_score for expert in self.human_experts])
            current_avg_trust = self.partnership_metrics['average_trust']
            self.partnership_metrics['average_trust'] = (current_avg_trust * (n-1) + avg_trust) / n
```

```python
def create_partnership_scenario():
    """Create human-AI partnership scenario for demonstration"""
    
    # Create AI agents
    ai_agents = [
        AIAgent(
            id="ai-001",
            name="OptimizationAI",
            capability="vessel_optimization",
            accuracy=0.85,
            explainability_score=0.75,
            confidence_threshold=0.8
        ),
        AIAgent(
            id="ai-002", 
            name="SafetyAI",
            capability="safety_compliance",
            accuracy=0.90,
            explainability_score=0.80,
            confidence_threshold=0.85
        )
    ]
    
    # Create human experts
    human_experts = [
        HumanExpert(
            id="he-001",
            name="Captain Chen",
            expertise_area="vessel_operations",
            experience_years=15,
            trust_score=0.6,
            confidence_level=0.8
        ),
        HumanExpert(
            id="he-002",
            name="Manager Kumar",
            expertise_area="port_operations",
            experience_years=12,
            trust_score=0.7,
            confidence_level=0.75
        ),
        HumanExpert(
            id="he-003",
            name="Inspector Smith",
            expertise_area="safety_compliance",
            experience_years=18,
            trust_score=0.65,
            confidence_level=0.85
        )
    ]
    
    # Create sample containers
    containers = []
    destinations = ["HKG", "SHA", "SIN", "KOR"]
    
    for i in range(12):
        container = type('Container', (), {
            'id': f'C{i+1:03d}',
            'destination_port': random.choice(destinations),
            'weight': random.uniform(15, 30),
            'current_position': (random.randint(1, 6), random.randint(1, 4), random.randint(1, 3)),
            'target_position': (random.randint(1, 6), random.randint(1, 4), random.randint(1, 3))
        })()
        containers.append(container)
    
    return ai_agents, human_experts, containers

# Create partnership scenario
ai_agents, human_experts, containers = create_partnership_scenario()

print(f"Partnership Configuration:")
print(f"  AI Agents: {len(ai_agents)}")
print(f"  Human Experts: {len(human_experts)}")
print(f"  Containers: {len(containers)}")
print(f"\nAI Agents:")
for agent in ai_agents:
    print(f"  {agent.name}: {agent.capability} (accuracy: {agent.accuracy:.1%})")
print(f"\nHuman Experts:")
for expert in human_experts:
    print(f"  {expert.name}: {expert.expertise_area} ({expert.experience_years} years, trust: {expert.trust_score:.1%})")
```

```python
# Initialize collaborative decision engine
print("\n=== INITIALIZING HUMAN-AI PARTNERSHIP ===")

decision_engine = CollaborativeDecisionEngine()

# Add AI agents and human experts
for agent in ai_agents:
    decision_engine.add_ai_agent(agent)

for expert in human_experts:
    decision_engine.add_human_expert(expert)

print(f"Partnership initialized with {len(decision_engine.ai_agents)} AI agents and {len(decision_engine.human_experts)} human experts")

# Simulate multiple decision scenarios
scenarios = [
    DecisionContext("SC-001", datetime.now(), 0.6, 0.7, 0.4, 0.5),  # Medium complexity
    DecisionContext("SC-002", datetime.now(), 0.8, 0.9, 0.6, 0.8),  # High complexity, critical
    DecisionContext("SC-003", datetime.now(), 0.4, 0.5, 0.3, 0.9),  # Low complexity, high time pressure
    DecisionContext("SC-004", datetime.now(), 0.7, 0.6, 0.8, 0.4),  # High uncertainty
    DecisionContext("SC-005", datetime.now(), 0.5, 0.8, 0.5, 0.6),  # Medium complexity, critical
]

collaborative_decisions = []

print(f"\n=== SIMULATING COLLABORATIVE DECISIONS ===")

for i, scenario in enumerate(scenarios, 1):
    print(f"\nScenario {i}: {scenario.problem_id}")
    print(f"  Complexity: {scenario.complexity_level:.1%}, Criticality: {scenario.criticality_level:.1%}")
    print(f"  Uncertainty: {scenario.uncertainty_level:.1%}, Time Pressure: {scenario.time_pressure:.1%}")
    
    # Make collaborative decision
    decision = decision_engine.make_collaborative_decision(containers, scenario)
    collaborative_decisions.append(decision)
    
    print(f"  Final Decision: {decision.final_decision['total_moves']} moves, ${decision.final_decision['total_cost']:.0f}")
    print(f"  Agreement Level: {decision.agreement_level:.1%}")
    print(f"  Decision Rationale: {decision.decision_rationale}")
    
    # Show trust updates
    if decision.trust_update:
        print(f"  Trust Updates:")
        for expert_id, change in decision.trust_update.items():
            expert = next(e for e in human_experts if e.id == expert_id)
            print(f"    {expert.name}: {change:+.3f} (new trust: {expert.trust_score:.1%})")
```

```python
# Analyze partnership performance
def analyze_partnership_performance(decisions: List[CollaborativeDecision]):
    """Analyze overall partnership performance"""
    
    print(f"\n=== PARTNERSHIP PERFORMANCE ANALYSIS ===")
    
    # Calculate metrics
    agreements = [d.agreement_level for d in decisions]
    moves = [d.final_decision['total_moves'] for d in decisions]
    costs = [d.final_decision['total_cost'] for d in decisions]
    
    print(f"\nDecision Quality Metrics:")
    print(f"  Average Agreement: {np.mean(agreements):.1%}")
    print(f"  Agreement Std Dev: {np.std(agreements):.1%}")
    print(f"  Average Moves: {np.mean(moves):.1f}")
    print(f"  Average Cost: ${np.mean(costs):.0f}")
    print(f"  Cost Std Dev: ${np.std(costs):.0f}")
    
    # Trust evolution
    print(f"\nTrust Evolution:")
    for expert in human_experts:
        print(f"  {expert.name}: {expert.trust_score:.1%} (started at {0.6:.1%} if vessel_operations else 0.65:.1%})")
    
    # Partnership metrics
    metrics = decision_engine.partnership_metrics
    print(f"\nPartnership Metrics:")
    print(f"  Total Decisions: {metrics['total_decisions']}")
    print(f"  Agreement Rate: {metrics['agreement_rate']:.1%}")
    print(f"  Average Trust: {metrics['average_trust']:.1%}")
    
    # Learning outcomes summary
    all_outcomes = []
    for decision in decisions:
        all_outcomes.extend(decision.learning_outcomes)
    
    print(f"\nKey Learning Outcomes:")
    unique_outcomes = list(set(all_outcomes))
    for outcome in unique_outcomes[:5]:  # Top 5
        print(f"  • {outcome}")
    
    return {
        'avg_agreement': np.mean(agreements),
        'avg_cost': np.mean(costs),
        'trust_scores': [expert.trust_score for expert in human_experts],
        'partnership_metrics': metrics
    }

# Analyze performance
performance_results = analyze_partnership_performance(collaborative_decisions)
```

```python
# Visualize partnership results
def visualize_partnership_results(decisions: List[CollaborativeDecision], performance: Dict):
    """Create comprehensive visualization of partnership results"""
    
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    
    # 1. Agreement level over time
    ax1 = axes[0, 0]
    agreements = [d.agreement_level for d in decisions]
    ax1.plot(range(1, len(agreements) + 1), agreements, 'bo-', linewidth=2, markersize=8)
    ax1.set_title('Human-AI Agreement Evolution', fontweight='bold')
    ax1.set_xlabel('Decision Number')
    ax1.set_ylabel('Agreement Level')
    ax1.set_ylim(0, 1)
    ax1.grid(True, alpha=0.3)
    
    # 2. Trust scores by expert
    ax2 = axes[0, 1]
    expert_names = [expert.name for expert in human_experts]
    trust_scores = [expert.trust_score for expert in human_experts]
    colors = ['green', 'blue', 'orange'][:len(expert_names)]
    
    bars = ax2.bar(expert_names, trust_scores, color=colors, alpha=0.7)
    ax2.set_title('Trust Scores in AI', fontweight='bold')
    ax2.set_ylabel('Trust Level')
    ax2.set_ylim(0, 1)
    
    # Add value labels on bars
    for bar, score in zip(bars, trust_scores):
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height + 0.02,
                f'{score:.1%}', ha='center', va='bottom', fontweight='bold')
    
    # 3. Decision costs comparison
    ax3 = axes[0, 2]
    costs = [d.final_decision['total_cost'] for d in decisions]
    moves = [d.final_decision['total_moves'] for d in decisions]
    
    ax3.scatter(moves, costs, c=agreements, cmap='viridis', s=100, alpha=0.7)
    ax3.set_title('Decision Cost vs Moves', fontweight='bold')
    ax3.set_xlabel('Number of Moves')
    ax3.set_ylabel('Total Cost ($)')
    
    # Add colorbar for agreement
    cbar = plt.colorbar(ax3.collections[0], ax=ax3)
    cbar.set_label('Agreement Level')
    
    # 4. Authority distribution (average across decisions)
    ax4 = axes[1, 0]
    
    # Calculate average authority distribution
    all_authorities = []
    for decision in decisions:
        all_authorities.append(decision.authority_distribution)
    
    if all_authorities:
        avg_authority = {}
        for key in all_authorities[0].keys():
            avg_authority[key] = np.mean([auth.get(key, 0) for auth in all_authorities])
        
        labels = list(avg_authority.keys())
        values = list(avg_authority.values())
        
        # Create pie chart
        ax4.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
        ax4.set_title('Average Authority Distribution', fontweight='bold')
    
    # 5. Learning outcomes frequency
    ax5 = axes[1, 1]
    
    # Count learning outcomes
    outcome_counts = {}
    for decision in decisions:
        for outcome in decision.learning_outcomes:
            # Shorten outcome for display
            short_outcome = outcome[:40] + "..." if len(outcome) > 40 else outcome
            outcome_counts[short_outcome] = outcome_counts.get(short_outcome, 0) + 1
    
    if outcome_counts:
        outcomes = list(outcome_counts.keys())
        counts = list(outcome_counts.values())
        
        ax5.barh(outcomes, counts, color='purple', alpha=0.7)
        ax5.set_title('Learning Outcomes Frequency', fontweight='bold')
        ax5.set_xlabel('Frequency')
    
    # 6. Partnership metrics dashboard
    ax6 = axes[1, 2]
    
    metrics = performance['partnership_metrics']
    metric_names = ['Agreement Rate', 'Average Trust', 'Total Decisions']
    metric_values = [metrics['agreement_rate'], metrics['average_trust'], metrics['total_decisions']/10]  # Scale decisions
    
    # Normalize values for display
    normalized_values = [v/max(metric_values) for v in metric_values]
    
    bars = ax6.bar(metric_names, normalized_values, color=['red', 'green', 'blue'], alpha=0.7)
    ax6.set_title('Partnership Health Metrics', fontweight='bold')
    ax6.set_ylabel('Normalized Score')
    ax6.set_ylim(0, 1)
    
    # Add actual values as text
    for i, (bar, actual_val) in enumerate(zip(bars, [metrics['agreement_rate'], metrics['average_trust'], metrics['total_decisions']])):
        height = bar.get_height()
        if i < 2:  # For percentages
            ax6.text(bar.get_x() + bar.get_width()/2., height + 0.02,
                    f'{actual_val:.1%}', ha='center', va='bottom', fontweight='bold')
        else:  # For count
            ax6.text(bar.get_x() + bar.get_width()/2., height + 0.02,
                    f'{int(actual_val)}', ha='center', va='bottom', fontweight='bold')
    
    plt.suptitle('Human-AI Symbiotic Partnership Analysis', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()

# Visualize results
visualize_partnership_results(collaborative_decisions, performance_results)
```
