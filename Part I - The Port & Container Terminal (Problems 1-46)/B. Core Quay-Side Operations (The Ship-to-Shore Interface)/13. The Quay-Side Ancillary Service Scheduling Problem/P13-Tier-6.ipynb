Building on all previous tiers (mathematical formulation, heuristics, metaheuristics, reinforcement learning, and digital twin), we now implement an **Autonomous Multi-Agent Ecosystem** - the most advanced approach that transforms ancillary service scheduling from centralized control to a decentralized network of intelligent agents that negotiate, collaborate, and self-optimize through emergent coordination mechanisms.

### Why This Tier Exists vs Previous Tiers:

**Limitations of Previous Approaches:**
- **Tier 1-3**: Centralized optimization, no autonomy or adaptability
- **Tier 4**: Single-agent learning, no multi-agent coordination
- **Tier 5**: System integration but centralized control

**Advantages of Multi-Agent Ecosystem:**
- **Decentralized Intelligence**: Each service operates as an autonomous agent
- **Emergent Optimization**: Global efficiency arises from local interactions
- **Game-Theoretic Coordination**: Agents negotiate using sophisticated strategies
- **Blockchain-Based Trust**: Smart contracts ensure transparent, enforceable agreements
- **Fault Tolerance**: System resilience through distributed decision making
- **Self-Organization**: No central controller required for coordination

### Multi-Agent Architecture Theory

The ecosystem models each ancillary service as an autonomous agent with:
- **Individual Objectives**: Service-specific goals and constraints
- **Negotiation Capabilities**: Auction-based and bargaining protocols
- **Learning Mechanisms**: Reinforcement learning for strategy adaptation
- **Reputation Systems**: Trust building through historical performance

Agents engage in continuous negotiations using game-theoretic strategies, creating emergent optimization where global efficiency arises from local rational decisions.

```python
class ServiceAgent:
    """
    Autonomous agent for ancillary service scheduling
    Each agent represents a service type with its own objectives and negotiation capabilities
    """
    
    def __init__(self, agent_id, service_type, capabilities, resources):
        self.agent_id = agent_id
        self.service_type = service_type
        self.capabilities = capabilities  # What the agent can do
        self.resources = resources  # Resources controlled by the agent
        
        # Agent state
        self.current_schedule = {}
        self.resource_availability = resources.copy()
        self.reputation_score = 1.0  # Initial reputation
        self.success_history = deque(maxlen=100)
        self.negotiation_history = deque(maxlen=50)
        
        # Learning parameters
        self.learning_rate = 0.1
        self.exploration_rate = 0.2
        self.strategy_weights = {
            'aggressive': 0.3,
            'cooperative': 0.4,
            'conservative': 0.3
        }
        
        # Objectives and constraints
        self.objectives = {
            'schedule_efficiency': 0.4,
            'resource_utilization': 0.3,
            'cost_minimization': 0.2,
            'service_quality': 0.1
        }
        
        # Negotiation strategy
        self.negotiation_strategy = 'cooperative'
        self.min_acceptable_utilization = 0.6
        self.max_bid_price = 1000
        
    def calculate_utility(self, schedule_slot, resource_requirements):
        """
        Calculate utility for a given schedule slot based on agent's objectives
        """
        utility = 0
        
        # Schedule efficiency component
        time_preference = 1.0 - abs(schedule_slot['time'] - 12) / 12  # Preference for mid-day
        utility += self.objectives['schedule_efficiency'] * time_preference
        
        # Resource utilization component
        resource_fit = 1.0
        for resource_type, required in resource_requirements.items():
            available = self.resource_availability.get(resource_type, 0)
            if available >= required:
                resource_fit *= (required / available)
            else:
                resource_fit *= 0.1  # Heavy penalty for insufficient resources
        utility += self.objectives['resource_utilization'] * resource_fit
        
        # Cost minimization component
        estimated_cost = schedule_slot.get('estimated_cost', 500)
        cost_score = 1.0 - (estimated_cost / self.max_bid_price)
        utility += self.objectives['cost_minimization'] * cost_score
        
        # Service quality component
        quality_factor = schedule_slot.get('quality_factor', 0.8)
        utility += self.objectives['service_quality'] * quality_factor
        
        return utility
    
    def make_bid(self, auction_item, current_round):
        """
        Make a bid in an auction for a schedule slot
        """
        resource_requirements = auction_item['resource_requirements']
        utility = self.calculate_utility(auction_item, resource_requirements)
        
        # Check if agent has sufficient resources
        can_fulfill = True
        for resource_type, required in resource_requirements.items():
            if self.resource_availability.get(resource_type, 0) < required:
                can_fulfill = False
                break
        
        if not can_fulfill or utility < self.min_acceptable_utilization:
            return None  # No bid
        
        # Calculate bid based on strategy
        base_price = auction_item.get('base_price', 500)
        
        if self.negotiation_strategy == 'aggressive':
            bid_price = base_price * (1.0 + utility * 0.5)
        elif self.negotiation_strategy == 'cooperative':
            bid_price = base_price * (0.8 + utility * 0.4)
        else:  # conservative
            bid_price = base_price * (0.6 + utility * 0.3)
        
        bid_price = min(bid_price, self.max_bid_price)
        
        return {
            'agent_id': self.agent_id,
            'service_type': self.service_type,
            'bid_price': bid_price,
            'utility': utility,
            'strategy': self.negotiation_strategy,
            'resource_commitment': resource_requirements.copy()
        }
    
    def update_strategy(self, negotiation_result):
        """
        Update negotiation strategy based on results using reinforcement learning
        """
        self.negotiation_history.append(negotiation_result)
        
        # Update reputation based on success
        if negotiation_result.get('success', False):
            self.reputation_score = min(1.0, self.reputation_score + 0.05)
            self.success_history.append(1)
        else:
            self.reputation_score = max(0.0, self.reputation_score - 0.02)
            self.success_history.append(0)
        
        # Learn from experience
        if len(self.success_history) >= 10:
            recent_success_rate = np.mean(list(self.success_history)[-10:])
            
            # Adjust strategy weights based on success
            if negotiation_result.get('strategy') == 'aggressive' and recent_success_rate < 0.3:
                self.strategy_weights['aggressive'] *= 0.9
                self.strategy_weights['cooperative'] *= 1.05
                self.strategy_weights['conservative'] *= 1.05
            
            # Normalize weights
            total_weight = sum(self.strategy_weights.values())
            for strategy in self.strategy_weights:
                self.strategy_weights[strategy] /= total_weight
            
            # Update negotiation strategy
            self.negotiation_strategy = random.choices(
                list(self.strategy_weights.keys()),
                weights=list(self.strategy_weights.values())
            )[0]
    
    def allocate_resources(self, resource_commitment):
        """
        Allocate resources for a committed schedule
        """
        for resource_type, amount in resource_commitment.items():
            if resource_type in self.resource_availability:
                self.resource_availability[resource_type] -= amount
    
    def release_resources(self, resource_commitment):
        """
        Release resources when a schedule is completed
        """
        for resource_type, amount in resource_commitment.items():
            if resource_type in self.resource_availability:
                self.resource_availability[resource_type] += amount
    
    def get_agent_status(self):
        """
        Get current agent status for monitoring
        """
        return {
            'agent_id': self.agent_id,
            'service_type': self.service_type,
            'reputation_score': self.reputation_score,
            'current_strategy': self.negotiation_strategy,
            'resource_availability': self.resource_availability.copy(),
            'active_schedules': len(self.current_schedule),
            'success_rate': np.mean(list(self.success_history)) if self.success_history else 0.0
        }

print("Service Agent class defined successfully!")
```

```python
class AuctionManager:
    """
    Manages auctions for schedule slots using different auction mechanisms
    """
    
    def __init__(self, auction_type='sealed_bid'):
        self.auction_type = auction_type  # 'sealed_bid', 'english', 'dutch'
        self.auction_history = deque(maxlen=100)
        self.market_clearing_price = None
        
    def run_sealed_bid_auction(self, auction_item, participating_agents):
        """
        Run a sealed-bid auction for a schedule slot
        """
        bids = []
        
        # Collect bids from all participating agents
        for agent in participating_agents:
            bid = agent.make_bid(auction_item, 1)
            if bid:
                bids.append(bid)
        
        if not bids:
            return {
                'success': False,
                'reason': 'No valid bids received'
            }
        
        # Sort bids by price (highest first)
        bids.sort(key=lambda x: x['bid_price'], reverse=True)
        
        # Winner determination
        winning_bid = bids[0]
        winner_agent = next(agent for agent in participating_agents 
                           if agent.agent_id == winning_bid['agent_id'])
        
        # Market clearing price (second-highest bid for efficiency)
        if len(bids) > 1:
            self.market_clearing_price = bids[1]['bid_price']
        else:
            self.market_clearing_price = winning_bid['bid_price']
        
        # Record auction result
        auction_result = {
            'auction_type': 'sealed_bid',
            'auction_item': auction_item,
            'winning_bid': winning_bid,
            'winner_agent_id': winning_bid['agent_id'],
            'market_clearing_price': self.market_clearing_price,
            'total_bids': len(bids),
            'success': True
        }
        
        self.auction_history.append(auction_result)
        
        return auction_result
    
    def run_english_auction(self, auction_item, participating_agents, max_rounds=10):
        """
        Run an English auction (ascending price)
        """
        current_price = auction_item.get('base_price', 100)
        price_increment = auction_item.get('price_increment', 50)
        active_agents = participating_agents.copy()
        auction_round = 0
        
        while len(active_agents) > 1 and auction_round < max_rounds:
            auction_round += 1
            
            # Get bids from active agents
            round_bids = []
            for agent in active_agents:
                bid = agent.make_bid({
                    **auction_item,
                    'base_price': current_price
                }, auction_round)
                
                if bid and bid['bid_price'] >= current_price:
                    round_bids.append(bid)
            
            # Filter agents who didn't bid high enough
            if round_bids:
                # Find highest bid
                highest_bid = max(round_bids, key=lambda x: x['bid_price'])
                current_price = highest_bid['bid_price'] + price_increment
                
                # Keep only agents who can meet the new price
                active_agents = [agent for agent in active_agents 
                                if any(bid['agent_id'] == agent.agent_id 
                                          for bid in round_bids 
                                          if bid['bid_price'] >= current_price - price_increment)]
            else:
                break
        
        # Determine winner
        if len(active_agents) == 1:
            winner_agent = active_agents[0]
            winning_bid = winner_agent.make_bid({
                **auction_item,
                'base_price': current_price - price_increment
            }, auction_round)
            
            auction_result = {
                'auction_type': 'english',
                'auction_item': auction_item,
                'winning_bid': winning_bid,
                'winner_agent_id': winner_agent.agent_id,
                'final_price': current_price - price_increment,
                'rounds': auction_round,
                'success': True
            }
        else:
            auction_result = {
                'success': False,
                'reason': 'No winner determined'
            }
        
        self.auction_history.append(auction_result)
        return auction_result

print("Auction Manager class defined successfully!")
```

```python
class BlockchainContractManager:
    """
    Manages smart contracts for service agreements and resource allocations
    Simulates blockchain functionality for transparency and enforcement
    """
    
    def __init__(self):
        self.contracts = {}  # contract_id -> contract_details
        self.contract_history = deque(maxlen=1000)
        self.ledger_balance = defaultdict(float)  # agent_id -> balance
        self.transaction_fee = 0.01  # 1% transaction fee
        
    def create_service_contract(self, contract_details):
        """
        Create a smart contract for service agreement
        """
        contract_id = f"CONTRACT_{len(self.contracts) + 1:06d}"
        
        contract = {
            'contract_id': contract_id,
            'service_provider': contract_details['provider_agent_id'],
            'service_requester': contract_details['requester_agent_id'],
            'service_type': contract_details['service_type'],
            'schedule_slot': contract_details['schedule_slot'],
            'resource_commitment': contract_details['resource_commitment'],
            'price': contract_details['price'],
            'penalty_clause': contract_details.get('penalty_clause', 0.1),  # 10% penalty
            'quality_requirements': contract_details.get('quality_requirements', 0.8),
            'created_at': datetime.now(),
            'status': 'active',
            'executed_at': None,
            'completion_verified': False
        }
        
        # Record contract on blockchain
        self.contracts[contract_id] = contract
        
        # Process payment
        payment_amount = contract['price']
        transaction_fee = payment_amount * self.transaction_fee
        
        # Deduct from requester
        self.ledger_balance[contract['service_requester']] -= (payment_amount + transaction_fee)
        # Hold in escrow
        self.ledger_balance[f"ESCROW_{contract_id}"] = payment_amount
        
        self.contract_history.append({
            'action': 'contract_created',
            'contract_id': contract_id,
            'timestamp': datetime.now()
        })
        
        return contract_id
    
    def execute_contract(self, contract_id, completion_quality=1.0):
        """
        Execute a completed contract and release payment
        """
        if contract_id not in self.contracts:
            return {'success': False, 'reason': 'Contract not found'}
        
        contract = self.contracts[contract_id]
        
        if contract['status'] != 'active':
            return {'success': False, 'reason': 'Contract not active'}
        
        # Verify quality requirements
        if completion_quality < contract['quality_requirements']:
            # Apply penalty
            penalty = contract['price'] * contract['penalty_clause']
            final_payment = contract['price'] - penalty
            
            # Release reduced payment
            self.ledger_balance[contract['service_provider']] += final_payment
            self.ledger_balance[f"ESCROW_{contract_id}"] = 0
            
            contract['status'] = 'completed_with_penalty'
            contract['completion_verified'] = True
            contract['executed_at'] = datetime.now()
            contract['actual_quality'] = completion_quality
            
            return {
                'success': True,
                'payment': final_payment,
                'penalty_applied': penalty,
                'quality_met': False
            }
        else:
            # Full payment
            self.ledger_balance[contract['service_provider']] += contract['price']
            self.ledger_balance[f"ESCROW_{contract_id}"] = 0
            
            contract['status'] = 'completed_successfully'
            contract['completion_verified'] = True
            contract['executed_at'] = datetime.now()
            contract['actual_quality'] = completion_quality
            
            return {
                'success': True,
                'payment': contract['price'],
                'penalty_applied': 0,
                'quality_met': True
            }
    
    def get_contract_status(self, contract_id):
        """
        Get contract status and details
        """
        if contract_id not in self.contracts:
            return {'error': 'Contract not found'}
        
        contract = self.contracts[contract_id]
        return {
            'contract_id': contract_id,
            'status': contract['status'],
            'provider': contract['service_provider'],
            'requester': contract['service_requester'],
            'price': contract['price'],
            'created_at': contract['created_at'],
            'executed_at': contract.get('executed_at'),
            'completion_verified': contract['completion_verified']
        }
    
    def get_agent_balance(self, agent_id):
        """
        Get agent's current balance
        """
        return self.ledger_balance.get(agent_id, 0.0)
    
    def get_market_statistics(self):
        """
        Get market statistics and contract performance
        """
        total_contracts = len(self.contracts)
        completed_contracts = sum(1 for c in self.contracts.values() 
                                if c['status'].startswith('completed'))
        
        if completed_contracts > 0:
            avg_completion_time = np.mean([
                (c['executed_at'] - c['created_at']).total_seconds() / 3600
                for c in self.contracts.values()
                if c['executed_at'] is not None
            ])
        else:
            avg_completion_time = 0
        
        return {
            'total_contracts': total_contracts,
            'completed_contracts': completed_contracts,
            'completion_rate': completed_contracts / total_contracts if total_contracts > 0 else 0,
            'average_completion_time_hours': avg_completion_time,
            'total_transaction_fees': total_contracts * 5  # Assuming avg price of 500
        }

print("Blockchain Contract Manager class defined successfully!")
```

```python
class AutonomousMultiAgentEcosystem:
    """
    Main multi-agent ecosystem for autonomous service scheduling
    Integrates agents, auctions, and blockchain contracts
    """
    
    def __init__(self):
        self.agents = {}
        self.auction_manager = AuctionManager()
        self.blockchain_manager = BlockchainContractManager()
        
        # System state
        self.current_time = datetime.now()
        self.schedule_slots = []
        self.active_contracts = {}
        
        # Performance metrics
        self.system_metrics = deque(maxlen=1000)
        self.coordination_events = deque(maxlen=500)
        
        # Initialize agents
        self.initialize_agents()
        
        # Initialize schedule slots
        self.initialize_schedule_slots()
        
        # Give agents initial balances
        self.initialize_agent_balances()
    
    def initialize_agents(self):
        """
        Initialize autonomous service agents
        """
        # Define agent types and their capabilities
        agent_configs = [
            {
                'agent_id': 'MAINT_001',
                'service_type': 'Maintenance',
                'capabilities': ['crane_maintenance', 'equipment_repair', 'inspection'],
                'resources': {'technicians': 8, 'supervisors': 2, 'tools': 15}
            },
            {
                'agent_id': 'SECURITY_001',
                'service_type': 'Security',
                'capabilities': ['patrol', 'access_control', 'surveillance'],
                'resources': {'guards': 12, 'vehicles': 4, 'equipment': 8}
            },
            {
                'agent_id': 'EMERGENCY_001',
                'service_type': 'Emergency',
                'capabilities': ['emergency_response', 'fire_suppression', 'medical_assistance'],
                'resources': {'paramedics': 6, 'vehicles': 3, 'equipment': 10}
            },
            {
                'agent_id': 'TUGBOAT_001',
                'service_type': 'Tugboat',
                'capabilities': ['vessel_positioning', 'escort', 'emergency_towing'],
                'resources': {'vessels': 3, 'crew': 9, 'fuel': 1000}
            },
            {
                'agent_id': 'POWER_001',
                'service_type': 'Power',
                'capabilities': ['backup_power', 'grid_maintenance', 'energy_management'],
                'resources': {'technicians': 5, 'equipment': 12, 'fuel': 500}
            },
            {
                'agent_id': 'ENVIRONMENT_001',
                'service_type': 'Environmental',
                'capabilities': ['pollution_monitoring', 'waste_management', 'compliance'],
                'resources': {'technicians': 4, 'equipment': 8, 'vehicles': 2}
            }
        ]
        
        # Create agents
        for config in agent_configs:
            agent = ServiceAgent(
                config['agent_id'],
                config['service_type'],
                config['capabilities'],
                config['resources']
            )
            self.agents[config['agent_id']] = agent
    
    def initialize_schedule_slots(self):
        """
        Initialize schedule slots for auction
        """
        # Create 24-hour schedule with 1-hour slots
        for hour in range(24):
            for slot_type in ['routine', 'priority', 'emergency']:
                slot = {
                    'slot_id': f"SLOT_{hour:02d}_{slot_type.upper()}",
                    'time': hour,
                    'type': slot_type,
                    'base_price': 200 if slot_type == 'routine' else 500 if slot_type == 'priority' else 1000,
                    'resource_requirements': self.generate_resource_requirements(slot_type),
                    'quality_factor': 0.8 if slot_type == 'routine' else 0.9 if slot_type == 'priority' else 0.95,
                    'duration': 1,  # hours
                    'allocated': False
                }
                self.schedule_slots.append(slot)
    
    def generate_resource_requirements(self, slot_type):
        """
        Generate resource requirements for different slot types
        """
        if slot_type == 'routine':
            return {
                'technicians': random.randint(2, 4),
                'supervisors': random.randint(0, 1),
                'equipment': random.randint(3, 6)
            }
        elif slot_type == 'priority':
            return {
                'technicians': random.randint(4, 7),
                'supervisors': random.randint(1, 2),
                'equipment': random.randint(6, 10)
            }
        else:  # emergency
            return {
                'technicians': random.randint(6, 10),
                'supervisors': random.randint(2, 3),
                'equipment': random.randint(8, 15)
            }
    
    def initialize_agent_balances(self):
        """
        Give agents initial balances for bidding
        """
        for agent_id in self.agents:
            self.blockchain_manager.ledger_balance[agent_id] = 10000  # Initial balance
    
    def run_auction_round(self, slot):
        """
        Run an auction for a specific schedule slot
        """
        if slot['allocated']:
            return None
        
        # Determine participating agents based on slot type
        participating_agents = []
        
        if slot['type'] == 'emergency':
            # Only emergency agents can bid
            participating_agents = [agent for agent in self.agents.values() 
                                 if agent.service_type in ['Emergency', 'Tugboat']]
        elif slot['type'] == 'priority':
            # Priority agents
            participating_agents = [agent for agent in self.agents.values() 
                                 if agent.service_type in ['Emergency', 'Tugboat', 'Security', 'Power']]
        else:  # routine
            # All agents can participate
            participating_agents = list(self.agents.values())
        
        # Run auction
        auction_result = self.auction_manager.run_sealed_bid_auction(slot, participating_agents)
        
        if auction_result['success']:
            # Create contract
            contract_details = {
                'provider_agent_id': auction_result['winner_agent_id'],
                'requester_agent_id': 'SYSTEM',  # System requests all services
                'service_type': slot['type'],
                'schedule_slot': slot,
                'resource_commitment': auction_result['winning_bid']['resource_commitment'],
                'price': auction_result['winning_bid']['bid_price'],
                'penalty_clause': 0.1,
                'quality_requirements': slot['quality_factor']
            }
            
            contract_id = self.blockchain_manager.create_service_contract(contract_details)
            
            # Update agent allocation
            winner_agent = self.agents[auction_result['winner_agent_id']]
            winner_agent.allocate_resources(auction_result['winning_bid']['resource_commitment'])
            
            # Mark slot as allocated
            slot['allocated'] = True
            slot['allocated_to'] = auction_result['winner_agent_id']
            slot['contract_id'] = contract_id
            
            # Store active contract
            self.active_contracts[contract_id] = {
                'contract_id': contract_id,
                'agent_id': auction_result['winner_agent_id'],
                'slot': slot,
                'execution_time': self.current_time + timedelta(hours=slot['time']),
                'status': 'pending'
            }
            
            # Record coordination event
            self.coordination_events.append({
                'event_type': 'auction_won',
                'agent_id': auction_result['winner_agent_id'],
                'slot_id': slot['slot_id'],
                'price': auction_result['winning_bid']['bid_price'],
                'timestamp': self.current_time
            })
            
            # Update agent strategy
            winner_agent.update_strategy({
                'success': True,
                'strategy': auction_result['winning_bid']['strategy'],
                'price': auction_result['winning_bid']['bid_price']
            })
            
            return {
                'success': True,
                'contract_id': contract_id,
                'winner': auction_result['winner_agent_id'],
                'price': auction_result['winning_bid']['bid_price']
            }
        else:
            return auction_result
    
    def execute_pending_contracts(self):
        """
        Execute contracts that are due
        """
        executed_contracts = []
        
        for contract_id, contract_info in list(self.active_contracts.items()):
            if contract_info['status'] == 'pending' and contract_info['execution_time'] <= self.current_time:
                # Execute contract
                agent = self.agents[contract_info['agent_id']]
                
                # Simulate service execution quality
                execution_quality = random.uniform(0.7, 1.0)
                
                # Execute contract on blockchain
                execution_result = self.blockchain_manager.execute_contract(contract_id, execution_quality)
                
                # Release resources
                slot = contract_info['slot']
                agent.release_resources(slot['resource_requirements'])
                
                # Update agent strategy based on execution
                agent.update_strategy({
                    'success': execution_result['success'],
                    'quality_met': execution_result['quality_met'],
                    'payment': execution_result['payment']
                })
                
                # Update contract status
                contract_info['status'] = 'executed'
                executed_contracts.append({
                    'contract_id': contract_id,
                    'agent_id': contract_info['agent_id'],
                    'execution_quality': execution_quality,
                    'payment': execution_result['payment'],
                    'quality_met': execution_result['quality_met']
                    'penalty_applied': execution_result['penalty_applied']
                })
                
                # Record coordination event
                self.coordination_events.append({
                    'event_type': 'contract_executed',
                    'contract_id': contract_id,
                    'agent_id': contract_info['agent_id'],
                    'quality': execution_quality,
                    'timestamp': self.current_time
                })
        
        return executed_contracts
    
    def run_simulation_step(self):
        """
        Run one step of the multi-agent simulation
        """
        # Advance time
        self.current_time += timedelta(hours=1)
        
        # Execute pending contracts
        executed_contracts = self.execute_pending_contracts()
        
        # Run auctions for unallocated slots
        auction_results = []
        for slot in self.schedule_slots:
            if not slot['allocated'] and slot['time'] <= self.current_time.hour:
                auction_result = self.run_auction_round(slot)
                if auction_result:
                    auction_results.append(auction_result)
        
        # Calculate system metrics
        metrics = self.calculate_system_metrics()
        self.system_metrics.append({
            'timestamp': self.current_time,
            'metrics': metrics
        })
        
        return {
            'timestamp': self.current_time,
            'executed_contracts': executed_contracts,
            'auction_results': auction_results,
            'system_metrics': metrics
        }
    
    def calculate_system_metrics(self):
        """
        Calculate system-wide performance metrics
        """
        # Agent performance
        agent_reputations = [agent.reputation_score for agent in self.agents.values()]
        avg_reputation = np.mean(agent_reputations)
        
        # Resource utilization
        total_resources = defaultdict(int)
        used_resources = defaultdict(int)
        
        for agent in self.agents.values():
            for resource_type, total in agent.resources.items():
                total_resources[resource_type] += total
                used_resources[resource_type] += (total - agent.resource_availability[resource_type])
        
        resource_utilization = {}
        for resource_type in total_resources:
            if total_resources[resource_type] > 0:
                resource_utilization[resource_type] = used_resources[resource_type] / total_resources[resource_type]
        
        avg_resource_utilization = np.mean(list(resource_utilization.values())) if resource_utilization else 0
        
        # Market efficiency
        market_stats = self.blockchain_manager.get_market_statistics()
        
        # Coordination effectiveness
        recent_events = [e for e in self.coordination_events 
                        if (self.current_time - e['timestamp']).total_seconds() < 3600]  # Last hour
        
        return {
            'avg_agent_reputation': avg_reputation,
            'avg_resource_utilization': avg_resource_utilization,
            'resource_utilization_by_type': resource_utilization,
            'contract_completion_rate': market_stats['completion_rate'],
            'total_contracts': market_stats['total_contracts'],
            'coordination_events_per_hour': len(recent_events),
            'active_agents': len([a for a in self.agents.values() if a.resource_availability != a.resources]),
            'market_efficiency': market_stats['completion_rate'] * avg_resource_utilization
        }
    
    def get_ecosystem_status(self):
        """
        Get comprehensive ecosystem status
        """
        agent_statuses = {agent_id: agent.get_agent_status() for agent_id, agent in self.agents.items()}
        
        return {
            'current_time': self.current_time,
            'total_agents': len(self.agents),
            'agent_statuses': agent_statuses,
            'active_contracts': len(self.active_contracts),
            'total_slots': len(self.schedule_slots),
            'allocated_slots': sum(1 for slot in self.schedule_slots if slot['allocated']),
            'market_stats': self.blockchain_manager.get_market_statistics()
            'recent_metrics': self.system_metrics[-1]['metrics'] if self.system_metrics else None
        }

print("Autonomous Multi-Agent Ecosystem class defined successfully!")
```
