# 13. The Quay-Side Ancillary Service Scheduling Problem
## Digital Twin System-of-Systems Simulation

### Problem Introduction

Building on the mathematical formulation (Tier 1), heuristic methods (Tiers 2-3), and reinforcement learning (Tier 4), we now implement a **Digital Twin System-of-Systems Simulation** - an advanced integrated approach that transforms ancillary service scheduling from isolated optimization to comprehensive system-wide simulation, enabling real-time synchronization between physical port operations and virtual decision-making processes.

### Why This Tier Exists vs Previous Tiers:

**Limitations of Previous Approaches:**
- **Tier 1-3**: Static optimization, no real-world integration
- **Tier 4**: Single-agent learning, no system-wide coordination
- **All Previous**: No consideration of interconnected systems and real-time data

**Advantages of Digital Twin Approach:**
- **Real-Time Integration**: Continuous bi-directional data exchange with physical systems
- **System-of-Systems View**: Comprehensive modeling of interconnected port subsystems
- **Predictive Analytics**: Forecast future conditions and optimize proactively
- **What-If Analysis**: Test scenarios before implementation
- **Continuous Learning**: System improves through operational data feedback

### Digital Twin Architecture Theory

The Digital Twin paradigm integrates multiple interconnected digital replicas:
- **Crane Operations Twin**: Real-time crane utilization and performance
- **Resource Availability Twin**: Equipment and personnel status tracking
- **Environmental Conditions Twin**: Weather and operational constraints
- **Service Performance Twin**: Historical service execution and quality metrics

These subsystems communicate through standardized APIs, creating a unified decision-making ecosystem that continuously learns and adapts to changing conditions.

```python
class IoTDataSimulator:
    """
    Simulates IoT sensor data from physical port operations
    Provides realistic data streams for digital twin integration
    """
    
    def __init__(self, num_sensors=150):
        self.num_sensors = num_sensors
        self.sensor_data = {}
        self.sensor_types = {
            'crane_utilization': 20,
            'equipment_status': 30,
            'personnel_location': 25,
            'weather_conditions': 15,
            'service_quality': 20,
            'traffic_flow': 25,
            'energy_consumption': 15
        }
        
        # Initialize sensors
        sensor_id = 0
        for sensor_type, count in self.sensor_types.items():
            for i in range(count):
                self.sensor_data[f"{sensor_type}_{sensor_id}"] = {
                    'type': sensor_type,
                    'id': sensor_id,
                    'location': random.choice(['Berth_A', 'Berth_B', 'Berth_C', 'Yard_1', 'Yard_2']),
                    'status': 'active',
                    'last_update': datetime.now(),
                    'value': self.generate_initial_value(sensor_type)
                }
                sensor_id += 1
    
    def generate_initial_value(self, sensor_type):
        """
        Generate realistic initial sensor values
        """
        if sensor_type == 'crane_utilization':
            return random.uniform(0.6, 0.95)  # 60-95% utilization
        elif sensor_type == 'equipment_status':
            return random.choice(['operational', 'maintenance', 'standby', 'error'])
        elif sensor_type == 'personnel_location':
            return {'x': random.uniform(0, 1000), 'y': random.uniform(0, 500)}
        elif sensor_type == 'weather_conditions':
            return {
                'temperature': random.uniform(10, 30),
                'wind_speed': random.uniform(0, 25),
                'visibility': random.uniform(1000, 10000),
                'precipitation': random.uniform(0, 10)
            }
        elif sensor_type == 'service_quality':
            return random.uniform(3.5, 5.0)  # Service rating
        elif sensor_type == 'traffic_flow':
            return random.randint(5, 50)  # Vehicles per hour
        elif sensor_type == 'energy_consumption':
            return random.uniform(100, 500)  # kWh
        
        return 0
    
    def update_sensor_data(self):
        """
        Update all sensor values with realistic changes
        """
        current_time = datetime.now()
        
        for sensor_id, sensor_info in self.sensor_data.items():
            sensor_type = sensor_info['type']
            
            # Update based on sensor type
            if sensor_type == 'crane_utilization':
                # Gradual changes with occasional spikes
                change = random.gauss(0, 0.02)
                new_value = sensor_info['value'] + change
                sensor_info['value'] = max(0, min(1, new_value))
            
            elif sensor_type == 'equipment_status':
                # Occasional status changes
                if random.random() < 0.05:  # 5% chance of status change
                    statuses = ['operational', 'maintenance', 'standby', 'error']
                    sensor_info['value'] = random.choice(statuses)
            
            elif sensor_type == 'personnel_location':
                # Movement simulation
                current_loc = sensor_info['value']
                new_x = current_loc['x'] + random.gauss(0, 10)
                new_y = current_loc['y'] + random.gauss(0, 5)
                sensor_info['value'] = {
                    'x': max(0, min(1000, new_x)),
                    'y': max(0, min(500, new_y))
                }
            
            elif sensor_type == 'weather_conditions':
                # Weather changes
                current = sensor_info['value']
                sensor_info['value'] = {
                    'temperature': current['temperature'] + random.gauss(0, 0.5),
                    'wind_speed': max(0, current['wind_speed'] + random.gauss(0, 1)),
                    'visibility': max(100, current['visibility'] + random.gauss(0, 100)),
                    'precipitation': max(0, current['precipitation'] + random.gauss(0, 0.5))
                }
            
            elif sensor_type == 'service_quality':
                # Service quality fluctuations
                change = random.gauss(0, 0.1)
                new_value = sensor_info['value'] + change
                sensor_info['value'] = max(1, min(5, new_value))
            
            elif sensor_type == 'traffic_flow':
                # Traffic flow variations
                change = random.randint(-5, 5)
                new_value = sensor_info['value'] + change
                sensor_info['value'] = max(0, min(100, new_value))
            
            elif sensor_type == 'energy_consumption':
                # Energy consumption patterns
                change = random.gauss(0, 10)
                new_value = sensor_info['value'] + change
                sensor_info['value'] = max(50, min(1000, new_value))
            
            sensor_info['last_update'] = current_time
    
    def get_aggregated_data(self):
        """
        Get aggregated data by sensor type and location
        """
        aggregated = {}
        
        for sensor_type in self.sensor_types.keys():
            type_sensors = [s for s in self.sensor_data.values() if s['type'] == sensor_type]
            
            if sensor_type == 'crane_utilization':
                values = [s['value'] for s in type_sensors]
                aggregated[sensor_type] = {
                    'average': np.mean(values),
                    'max': np.max(values),
                    'min': np.min(values),
                    'count': len(values)
                }
            
            elif sensor_type == 'equipment_status':
                statuses = [s['value'] for s in type_sensors]
                status_counts = {}
                for status in statuses:
                    status_counts[status] = status_counts.get(status, 0) + 1
                aggregated[sensor_type] = status_counts
            
            elif sensor_type == 'weather_conditions':
                # Average weather conditions
                avg_weather = {
                    'temperature': np.mean([s['value']['temperature'] for s in type_sensors]),
                    'wind_speed': np.mean([s['value']['wind_speed'] for s in type_sensors]),
                    'visibility': np.mean([s['value']['visibility'] for s in type_sensors]),
                    'precipitation': np.mean([s['value']['precipitation'] for s in type_sensors])
                }
                aggregated[sensor_type] = avg_weather
            
            else:
                values = [s['value'] for s in type_sensors if isinstance(s['value'], (int, float))]
                if values:
                    aggregated[sensor_type] = {
                        'average': np.mean(values),
                        'count': len(values)
                    }
        
        return aggregated

print("IoT Data Simulator class defined successfully!")
```

```python
class DigitalTwinSubsystem:
    """
    Base class for digital twin subsystems
    Each subsystem models a specific aspect of port operations
    """
    
    def __init__(self, name, subsystem_id):
        self.name = name
        self.subsystem_id = subsystem_id
        self.state = {}
        self.history = deque(maxlen=1000)  # Keep last 1000 states
        self.last_update = datetime.now()
        self.connected_subsystems = []
        
    def connect_to(self, other_subsystem):
        """
        Establish connection to another subsystem
        """
        if other_subsystem not in self.connected_subsystems:
            self.connected_subsystems.append(other_subsystem)
            other_subsystem.connected_subsystems.append(self)
    
    def update_state(self, new_state):
        """
        Update subsystem state and record in history
        """
        self.state.update(new_state)
        self.state['timestamp'] = datetime.now()
        self.history.append(self.state.copy())
        self.last_update = datetime.now()
    
    def get_state(self):
        """
        Get current subsystem state
        """
        return self.state.copy()
    
    def send_data(self, data, target_subsystem=None):
        """
        Send data to connected subsystems
        """
        if target_subsystem and target_subsystem in self.connected_subsystems:
            target_subsystem.receive_data(data, self)
        else:
            # Broadcast to all connected subsystems
            for subsystem in self.connected_subsystems:
                subsystem.receive_data(data, self)
    
    def receive_data(self, data, source_subsystem):
        """
        Receive data from another subsystem
        """
        # Process incoming data (to be overridden by subclasses)
        pass

class CraneOperationsTwin(DigitalTwinSubsystem):
    """
    Digital twin for crane operations and utilization
    """
    
    def __init__(self, num_cranes=6):
        super().__init__("Crane Operations", "CRANE_001")
        self.num_cranes = num_cranes
        self.crane_status = {}
        
        # Initialize crane status
        for i in range(num_cranes):
            self.crane_status[f"Crane_{i+1}"] = {
                'status': random.choice(['operational', 'maintenance', 'standby']),
                'utilization': random.uniform(0.6, 0.9),
                'current_task': random.choice(['loading', 'unloading', 'idle', 'maintenance']),
                'location': random.choice(['Berth_A', 'Berth_B', 'Berth_C']),
                'efficiency': random.uniform(0.8, 1.0)
            }
        
        self.update_state({
            'total_cranes': num_cranes,
            'operational_cranes': sum(1 for c in self.crane_status.values() if c['status'] == 'operational'),
            'average_utilization': np.mean([c['utilization'] for c in self.crane_status.values()]),
            'crane_details': self.crane_status.copy()
        })
    
    def update_from_iot_data(self, iot_data):
        """
        Update crane operations based on IoT sensor data
        """
        if 'crane_utilization' in iot_data:
            crane_data = iot_data['crane_utilization']
            
            # Update individual crane utilization
            for crane_id, crane_info in self.crane_status.items():
                # Simulate utilization changes
                change = random.gauss(0, 0.05)
                new_utilization = crane_info['utilization'] + change
                crane_info['utilization'] = max(0, min(1, new_utilization))
            
            # Update aggregated state
            self.update_state({
                'average_utilization': np.mean([c['utilization'] for c in self.crane_status.values()]),
                'operational_cranes': sum(1 for c in self.crane_status.values() if c['status'] == 'operational'),
                'crane_details': self.crane_status.copy()
            })
    
    def predict_maintenance_windows(self, hours_ahead=24):
        """
        Predict optimal maintenance windows based on current utilization patterns
        """
        current_utilization = self.state['average_utilization']
        
        # Simple prediction based on historical patterns
        maintenance_windows = []
        
        # Look for low utilization periods
        for hour in range(24):
            # Simulate utilization prediction (simplified)
            predicted_util = current_utilization + 0.2 * np.sin(hour * np.pi / 12)  # Daily pattern
            
            if predicted_util < 0.7:  # Low utilization threshold
                maintenance_windows.append({
                    'hour': hour,
                    'predicted_utilization': predicted_util,
                    'suitability': 'high' if predicted_util < 0.5 else 'medium'
                })
        
        return maintenance_windows

print("Digital Twin Subsystem classes defined successfully!")
```

```python
try:
    class ServiceSchedulingDigitalTwin:
        """
        Main Digital Twin system for service scheduling
        Integrates multiple subsystems for comprehensive simulation
        """
        
        def __init__(self):
            self.subsystems = {}
            self.iot_simulator = IoTDataSimulator(num_sensors=150)
            self.simulation_time = datetime.now()
            self.simulation_speed = 1  # 1 second = 1 minute in simulation
            
            # Performance metrics
            self.kpi_history = deque(maxlen=1000)
            self.alert_history = deque(maxlen=100)
            
            # Initialize subsystems
            self.initialize_subsystems()
            
            # Connect subsystems
            self.connect_subsystems()
        
        def initialize_subsystems(self):
            """
            Initialize all digital twin subsystems
            """
            # Crane Operations Twin
            self.subsystems['crane_operations'] = CraneOperationsTwin(num_cranes=6)
            
            # Resource Availability Twin
            self.subsystems['resource_availability'] = DigitalTwinSubsystem("Resource Availability", "RESOURCE_001")
            self.subsystems['resource_availability'].update_state({
                'technicians_available': 15,
                'supervisors_available': 6,
                'equipment_ready': 25,
                'availability_rate': 0.85
            })
            
            # Environmental Conditions Twin
            self.subsystems['environmental'] = DigitalTwinSubsystem("Environmental Conditions", "ENV_001")
            self.subsystems['environmental'].update_state({
                'weather': {
                    'temperature': 20.0,
                    'wind_speed': 10.0,
                    'visibility': 5000.0,
                    'precipitation': 0.0
                },
                'operational_constraints': []
            })
            
            # Service Performance Twin
            self.subsystems['service_performance'] = DigitalTwinSubsystem("Service Performance", "SERVICE_001")
            self.subsystems['service_performance'].update_state({
                'active_services': {},
                'completed_services': deque(maxlen=100),
                'service_quality_metrics': {
                    'average_rating': 4.2,
                    'completion_rate': 0.95,
                    'on_time_completion': 0.88
                }
            })
        
        def connect_subsystems(self):
            """
            Establish connections between subsystems
            """
            subsystems_list = list(self.subsystems.values())
            
            # Connect all subsystems to each other (full mesh)
            for i, subsystem1 in enumerate(subsystems_list):
                for subsystem2 in subsystems_list[i+1:]:
                    subsystem1.connect_to(subsystem2)
        
        def run_simulation_step(self):
            """
            Run one step of the digital twin simulation
            """
            # Update simulation time
            self.simulation_time += timedelta(minutes=5)  # 5-minute steps
            
            # Update IoT data
            self.iot_simulator.update_sensor_data()
            iot_data = self.iot_simulator.get_aggregated_data()
            
            # Update subsystems with IoT data
            for subsystem_name, subsystem in self.subsystems.items():
                if hasattr(subsystem, 'update_from_iot_data'):
                    subsystem.update_from_iot_data(iot_data)
            
            # Calculate system KPIs
            kpis = self.calculate_system_kpis()
            self.kpi_history.append({
                'timestamp': self.simulation_time,
                'kpis': kpis
            })
            
            return {
                'timestamp': self.simulation_time,
                'iot_data': iot_data,
                'kpis': kpis
            }
        
        def calculate_system_kpis(self):
            """
            Calculate system-wide Key Performance Indicators
            """
            crane_state = self.subsystems['crane_operations'].get_state()
            resource_state = self.subsystems['resource_availability'].get_state()
            service_state = self.subsystems['service_performance'].get_state()
            
            kpis = {
                'crane_utilization': crane_state['average_utilization'],
                'operational_cranes': crane_state['operational_cranes'],
                'resource_availability_rate': resource_state['availability_rate'],
                'active_services': len(service_state['active_services']),
                'system_efficiency': self.calculate_system_efficiency()
                'predictive_maintenance_opportunities': len(self.subsystems['crane_operations'].predict_maintenance_windows())
            }
            
            return kpis
        
        def calculate_system_efficiency(self):
            """
            Calculate overall system efficiency
            """
            crane_state = self.subsystems['crane_operations'].get_state()
            resource_state = self.subsystems['resource_availability'].get_state()
            
            # Simple efficiency calculation
            utilization_score = 1.0 - abs(crane_state['average_utilization'] - 0.8)  # Optimal around 80%
            availability_score = resource_state['availability_rate']
            
            overall_efficiency = (utilization_score + availability_score) / 2
            
            return overall_efficiency
    
    print("Digital Twin main class defined successfully!")
except Exception as e:
    print(f'Error in cell: {e}')
```
