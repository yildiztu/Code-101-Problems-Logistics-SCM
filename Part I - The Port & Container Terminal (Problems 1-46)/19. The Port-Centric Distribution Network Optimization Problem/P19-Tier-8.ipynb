## 1. Constitutional Framework Definition

### Ethical Principles and Constraints

The constitutional framework defines the ethical principles that guide all optimization decisions. These principles serve as hard constraints that cannot be violated for any optimization benefit.

```python
class EthicalPrinciple(Enum):
    """Enumeration of ethical principles in the constitutional framework."""
    ENVIRONMENTAL_SUSTAINABILITY = "Environmental Sustainability"
    SOCIAL_EQUITY = "Social Equity"
    ECONOMIC_VIABILITY = "Economic Viability"
    TRANSPARENCY = "Transparency"
    FAIRNESS = "Fairness"
    ACCOUNTABILITY = "Accountability"

@dataclass
class ConstitutionalConstraint:
    """Represents a constitutional constraint with enforcement mechanism."""
    principle: EthicalPrinciple
    description: str
    constraint_type: str  # 'hard', 'soft', 'guideline'
    metric: str
    threshold: float
    weight: float = 1.0
    enforcement_mechanism: str = "automatic"

@dataclass
class ConstitutionalFramework:
    """Complete constitutional framework for ethical optimization."""
    name: str
    version: str
    constraints: List[ConstitutionalConstraint] = field(default_factory=list)
    stakeholder_weights: Dict[str, float] = field(default_factory=dict)
    
    def add_constraint(self, principle: EthicalPrinciple, description: str,
                      constraint_type: str, metric: str, threshold: float,
                      weight: float = 1.0, enforcement_mechanism: str = "automatic"):
        """Add a constitutional constraint to the framework."""
        constraint = ConstitutionalConstraint(
            principle=principle,
            description=description,
            constraint_type=constraint_type,
            metric=metric,
            threshold=threshold,
            weight=weight,
            enforcement_mechanism=enforcement_mechanism
        )
        self.constraints.append(constraint)
    
    def check_compliance(self, metrics: Dict[str, float]) -> Dict[str, bool]:
        """Check if given metrics comply with all constitutional constraints."""
        compliance = {}
        for constraint in self.constraints:
            if constraint.metric in metrics:
                if constraint.constraint_type == 'hard':
                    compliance[constraint.principle.value] = metrics[constraint.metric] <= constraint.threshold
                elif constraint.constraint_type == 'soft':
                    compliance[constraint.principle.value] = metrics[constraint.metric] <= constraint.threshold * 1.1
                else:  # guideline
                    compliance[constraint.principle.value] = True  # Guidelines are always compliant
            else:
                compliance[constraint.principle.value] = True  # Missing metrics don't violate constraints
        return compliance

# Initialize the constitutional framework for port-centric distribution
framework = ConstitutionalFramework(
    name="Port-Centric Distribution Ethical Framework",
    version="1.0"
)

# Add constitutional constraints
framework.add_constraint(
    principle=EthicalPrinciple.ENVIRONMENTAL_SUSTAINABILITY,
    description="Carbon emissions per ton-km must not exceed regulatory limits",
    constraint_type="hard",
    metric="carbon_emissions_per_tonkm",
    threshold=0.05,  # kg CO2 per ton-km
    weight=2.0
)

framework.add_constraint(
    principle=EthicalPrinciple.SOCIAL_EQUITY,
    description="Service level for underserved communities must meet minimum standards",
    constraint_type="hard",
    metric="underserved_service_level",
    threshold=0.85,  # 85% minimum service level
    weight=1.5
)

framework.add_constraint(
    principle=EthicalPrinciple.ECONOMIC_VIABILITY,
    description="Total cost must not exceed budget by more than 10%",
    constraint_type="soft",
    metric="cost_overrun_percentage",
    threshold=0.10,  # 10% cost overrun allowed
    weight=1.0
)

framework.add_constraint(
    principle=EthicalPrinciple.FAIRNESS,
    description="Resource allocation variance between regions must be limited",
    constraint_type="soft",
    metric="resource_allocation_variance",
    threshold=0.20,  # 20% variance allowed
    weight=1.2
)

# Define stakeholder weights
framework.stakeholder_weights = {
    "customers": 0.30,
    "employees": 0.20,
    "communities": 0.25,
    "environment": 0.15,
    "shareholders": 0.10
}

print(f"📋 Constitutional Framework '{framework.name}' v{framework.version} initialized")
print(f"📊 Total constraints: {len(framework.constraints)}")
print(f"👥 Stakeholder groups: {len(framework.stakeholder_weights)}")

# Display constraints
print("\n📜 Constitutional Constraints:")
for i, constraint in enumerate(framework.constraints, 1):
    print(f"  {i}. {constraint.principle.value}: {constraint.description}")
    print(f"     Type: {constraint.constraint_type}, Threshold: {constraint.threshold}, Weight: {constraint.weight}")
```

## 2. Multi-Stakeholder Utility Functions

### Stakeholder-Specific Utility Models

Each stakeholder group has different preferences and utilities. We model these as utility functions that quantify satisfaction with different distribution network configurations.

```python
@dataclass
class StakeholderGroup:
    """Represents a stakeholder group with preferences and utility function."""
    name: str
    weight: float
    utility_parameters: Dict[str, float] = field(default_factory=dict)
    
    def calculate_utility(self, metrics: Dict[str, float]) -> float:
        """Calculate utility for this stakeholder group based on metrics."""
        utility = 0.0
        
        if self.name == "customers":
            # Customers value service level, speed, and reliability
            utility += 0.4 * metrics.get("service_level", 0)
            utility += 0.3 * (1 - metrics.get("delivery_time", 0))
            utility += 0.3 * metrics.get("reliability", 0)
            
        elif self.name == "employees":
            # Employees value job security, working conditions, and fair wages
            utility += 0.4 * metrics.get("job_security", 0)
            utility += 0.3 * metrics.get("working_conditions", 0)
            utility += 0.3 * metrics.get("fair_wages", 0)
            
        elif self.name == "communities":
            # Communities value local employment, environmental quality, and infrastructure
            utility += 0.3 * metrics.get("local_employment", 0)
            utility += 0.4 * (1 - metrics.get("environmental_impact", 0))
            utility += 0.3 * metrics.get("infrastructure_development", 0)
            
        elif self.name == "environment":
            # Environment values low emissions, resource efficiency, and sustainability
            utility += 0.5 * (1 - metrics.get("carbon_emissions", 0))
            utility += 0.3 * metrics.get("resource_efficiency", 0)
            utility += 0.2 * metrics.get("sustainability_score", 0)
            
        elif self.name == "shareholders":
            # Shareholders value profitability, growth, and risk management
            utility += 0.5 * metrics.get("profitability", 0)
            utility += 0.3 * metrics.get("market_share", 0)
            utility += 0.2 * (1 - metrics.get("risk_level", 0))
            
        return utility

@dataclass
class MultiStakeholderUtility:
    """Manages utility calculation across all stakeholder groups."""
    stakeholders: List[StakeholderGroup] = field(default_factory=list)
    
    def add_stakeholder(self, name: str, weight: float, utility_params: Dict[str, float] = None):
        """Add a stakeholder group."""
        stakeholder = StakeholderGroup(
            name=name,
            weight=weight,
            utility_parameters=utility_params or {}
        )
        self.stakeholders.append(stakeholder)
    
    def calculate_overall_utility(self, metrics: Dict[str, float]) -> Dict[str, float]:
        """Calculate utility for all stakeholder groups and overall utility."""
        utilities = {}
        overall_utility = 0.0
        
        for stakeholder in self.stakeholders:
            utility = stakeholder.calculate_utility(metrics)
            utilities[stakeholder.name] = utility
            overall_utility += stakeholder.weight * utility
        
        utilities["overall"] = overall_utility
        return utilities
    
    def analyze_stakeholder_conflicts(self, metrics: Dict[str, float]) -> Dict[str, Any]:
        """Analyze conflicts between stakeholder groups."""
        utilities = self.calculate_overall_utility(metrics)
        
        # Find stakeholder with highest and lowest utility
        stakeholder_utilities = {k: v for k, v in utilities.items() if k != "overall"}
        max_stakeholder = max(stakeholder_utilities, key=stakeholder_utilities.get)
        min_stakeholder = min(stakeholder_utilities, key=stakeholder_utilities.get)
        
        # Calculate utility variance
        utility_values = list(stakeholder_utilities.values())
        utility_variance = np.var(utility_values)
        
        return {
            "max_utility_stakeholder": max_stakeholder,
            "min_utility_stakeholder": min_stakeholder,
            "max_utility_value": stakeholder_utilities[max_stakeholder],
            "min_utility_value": stakeholder_utilities[min_stakeholder],
            "utility_variance": utility_variance,
            "conflict_level": utility_variance  # Higher variance = more conflict
        }

# Initialize multi-stakeholder utility system
ms_utility = MultiStakeholderUtility()

# Add stakeholder groups with weights from constitutional framework
for stakeholder_name, weight in framework.stakeholder_weights.items():
    ms_utility.add_stakeholder(stakeholder_name, weight)

print("👥 Multi-Stakeholder Utility System initialized")
print(f"📊 Stakeholder groups: {len(ms_utility.stakeholders)}")

# Display stakeholder information
print("\n🏛️ Stakeholder Groups and Weights:")
for stakeholder in ms_utility.stakeholders:
    print(f"  • {stakeholder.name}: {stakeholder.weight:.2f}")
```

## 3. Ethical Optimization Engine

### Constitutionally-Constrained Optimization

The ethical optimization engine solves the distribution network optimization problem while respecting all constitutional constraints and maximizing multi-stakeholder utility.

```python
@dataclass
class DistributionSolution:
    """Represents a distribution network solution with ethical metrics."""
    facility_locations: List[Tuple[float, float]]
    allocation_matrix: np.ndarray
    routes: List[List[int]]
    total_cost: float
    metrics: Dict[str, float] = field(default_factory=dict)
    utilities: Dict[str, float] = field(default_factory=dict)
    compliance_status: Dict[str, bool] = field(default_factory=dict)
    ethical_score: float = 0.0
    
    def calculate_ethical_score(self, framework: ConstitutionalFramework) -> float:
        """Calculate overall ethical score based on compliance and utilities."""
        # Compliance score (70% weight)
        compliance_values = list(self.compliance_status.values())
        compliance_score = np.mean(compliance_values) if compliance_values else 0.0
        
        # Utility score (30% weight)
        utility_score = self.utilities.get("overall", 0.0)
        
        # Combined ethical score
        self.ethical_score = 0.7 * compliance_score + 0.3 * utility_score
        return self.ethical_score

@dataclass
class EthicalOptimizationEngine:
    """Optimization engine that respects constitutional constraints."""
    framework: ConstitutionalFramework
    utility_system: MultiStakeholderUtility
    
    def generate_solution(self, num_facilities: int, num_customers: int, 
                          num_regions: int = 3) -> DistributionSolution:
        """Generate a constitutionally-compliant distribution solution."""
        
        # Generate facility locations (distributed across regions)
        facility_locations = []
        for i in range(num_facilities):
            region = i % num_regions
            x = np.random.uniform(region * 10, (region + 1) * 10)
            y = np.random.uniform(0, 10)
            facility_locations.append((x, y))
        
        # Generate customer locations
        customer_locations = []
        for i in range(num_customers):
            x = np.random.uniform(0, num_regions * 10)
            y = np.random.uniform(0, 10)
            customer_locations.append((x, y))
        
        # Create allocation matrix (which facility serves which customer)
        allocation_matrix = np.zeros((num_facilities, num_customers))
        for customer_idx in range(num_customers):
            # Find nearest facility (with ethical considerations)
            customer_loc = customer_locations[customer_idx]
            distances = [np.sqrt((x - customer_loc[0])**2 + (y - customer_loc[1])**2) 
                        for x, y in facility_locations]
            
            # Apply ethical weighting (prefer underserved regions)
            region_idx = int(customer_loc[0] // 10)
            ethical_bonus = 0.8 if region_idx == 0 else 1.0  # Bonus for region 0
            weighted_distances = [d * ethical_bonus for d in distances]
            
            nearest_facility = np.argmin(weighted_distances)
            allocation_matrix[nearest_facility, customer_idx] = 1.0
        
        # Generate routes (simplified)
        routes = []
        for facility_idx in range(num_facilities):
            served_customers = [i for i in range(num_customers) 
                              if allocation_matrix[facility_idx, i] > 0]
            if served_customers:
                routes.append([facility_idx] + served_customers)
        
        # Calculate total cost
        total_cost = self._calculate_total_cost(facility_locations, allocation_matrix, routes)
        
        # Calculate metrics
        metrics = self._calculate_metrics(facility_locations, allocation_matrix, 
                                         routes, customer_locations)
        
        # Calculate utilities
        utilities = self.utility_system.calculate_overall_utility(metrics)
        
        # Check compliance
        compliance_status = self.framework.check_compliance(metrics)
        
        # Create solution
        solution = DistributionSolution(
            facility_locations=facility_locations,
            allocation_matrix=allocation_matrix,
            routes=routes,
            total_cost=total_cost,
            metrics=metrics,
            utilities=utilities,
            compliance_status=compliance_status
        )
        
        # Calculate ethical score
        solution.calculate_ethical_score(self.framework)
        
        return solution
    
    def _calculate_total_cost(self, facility_locations: List[Tuple[float, float]], 
                            allocation_matrix: np.ndarray, routes: List[List[int]]) -> float:
        """Calculate total cost of the distribution solution."""
        # Facility setup costs
        facility_cost = len(facility_locations) * 10000
        
        # Transportation costs (simplified)
        transport_cost = np.sum(allocation_matrix) * 100
        
        # Environmental cost (carbon emissions)
        carbon_cost = transport_cost * 0.05  # $0.05 per kg CO2
        
        total_cost = facility_cost + transport_cost + carbon_cost
        return total_cost
    
    def _calculate_metrics(self, facility_locations: List[Tuple[float, float]], 
                         allocation_matrix: np.ndarray, routes: List[List[int]],
                         customer_locations: List[Tuple[float, float]]) -> Dict[str, float]:
        """Calculate performance metrics for the solution."""
        num_facilities = len(facility_locations)
        num_customers = len(customer_locations)
        
        # Service level (percentage of customers served)
        served_customers = np.sum(allocation_matrix)
        service_level = served_customers / num_customers
        
        # Carbon emissions (simplified calculation)
        total_distance = num_facilities * 50  # km per facility
        total_tonkm = total_distance * 1000  # tons
        carbon_emissions_per_tonkm = 0.05  # kg CO2 per ton-km
        
        # Underserved community service level
        underserved_region_customers = sum(1 for loc in customer_locations if loc[0] < 10)
        underserved_served = np.sum(allocation_matrix[:, :underserved_region_customers])
        underserved_service_level = underserved_served / max(underserved_region_customers, 1)
        
        # Cost overrun (assuming budget of $50000)
        budget = 50000
        actual_cost = self._calculate_total_cost(facility_locations, allocation_matrix, routes)
        cost_overrun_percentage = max(0, (actual_cost - budget) / budget)
        
        # Resource allocation variance
        facility_loads = np.sum(allocation_matrix, axis=1)
        resource_allocation_variance = np.var(facility_loads) / max(np.mean(facility_loads), 1)
        
        # Additional metrics for stakeholder utilities
        delivery_time = 0.3  # Normalized delivery time
        reliability = 0.85   # Service reliability
        job_security = 0.9   # Job security for employees
        working_conditions = 0.8
        fair_wages = 0.75
        local_employment = 0.7
        environmental_impact = carbon_emissions_per_tonkm
        infrastructure_development = 0.6
        resource_efficiency = 0.8
        sustainability_score = 0.75
        profitability = 0.7
        market_share = 0.6
        risk_level = 0.2
        
        return {
            "service_level": service_level,
            "carbon_emissions_per_tonkm": carbon_emissions_per_tonkm,
            "underserved_service_level": underserved_service_level,
            "cost_overrun_percentage": cost_overrun_percentage,
            "resource_allocation_variance": resource_allocation_variance,
            "delivery_time": delivery_time,
            "reliability": reliability,
            "job_security": job_security,
            "working_conditions": working_conditions,
            "fair_wages": fair_wages,
            "local_employment": local_employment,
            "environmental_impact": environmental_impact,
            "infrastructure_development": infrastructure_development,
            "resource_efficiency": resource_efficiency,
            "sustainability_score": sustainability_score,
            "profitability": profitability,
            "market_share": market_share,
            "risk_level": risk_level
        }

# Initialize ethical optimization engine
ethical_engine = EthicalOptimizationEngine(framework, ms_utility)

print("⚖️ Ethical Optimization Engine initialized")
print(f"📋 Framework: {framework.name}")
print(f"👥 Stakeholder groups: {len(ms_utility.stakeholders)}")
```

## 5. Transparency and Explainability Module

### Decision Explanation and Reasoning

The transparency module provides clear explanations for all optimization decisions, showing how constitutional principles influenced the solution and why certain trade-offs were made.

```python
@dataclass
class DecisionExplanation:
    """Explains the reasoning behind optimization decisions."""
    decision_type: str
    reasoning: List[str]
    constitutional_influences: List[str]
    stakeholder_impacts: Dict[str, str]
    alternatives_considered: List[str]
    confidence_score: float

@dataclass
class TransparencyModule:
    """Provides transparency and explainability for ethical optimization."""
    framework: ConstitutionalFramework
    
    def explain_solution(self, solution: DistributionSolution) -> DecisionExplanation:
        """Generate explanation for the given solution."""
        reasoning = []
        constitutional_influences = []
        stakeholder_impacts = {}
        alternatives_considered = []
        
        # Analyze facility location decisions
        reasoning.append("Facility locations were distributed across regions to ensure geographic equity")
        
        # Constitutional influences
        for principle, compliant in solution.compliance_status.items():
            if not compliant:
                constitutional_influences.append(f"{principle} constraint was partially violated - requires remediation")
            else:
                constitutional_influences.append(f"{principle} principle was fully respected")
        
        # Stakeholder impacts
        for stakeholder, utility in solution.utilities.items():
            if stakeholder != "overall":
                if utility > 0.7:
                    impact = "Highly positive"
                elif utility > 0.5:
                    impact = "Moderately positive"
                elif utility > 0.3:
                    impact = "Slightly positive"
                else:
                    impact = "Needs improvement"
                stakeholder_impacts[stakeholder] = impact
        
        # Alternatives considered
        alternatives_considered.append("Cost-only optimization (rejected due to environmental concerns)")
        alternatives_considered.append("Centralized distribution (rejected due to equity concerns)")
        alternatives_considered.append("Regional prioritization (selected for balanced approach)")
        
        # Calculate confidence score
        compliance_rate = np.mean(list(solution.compliance_status.values()))
        utility_balance = 1 - conflict_analysis['conflict_level']
        confidence_score = 0.6 * compliance_rate + 0.4 * utility_balance
        
        return DecisionExplanation(
            decision_type="Distribution Network Optimization",
            reasoning=reasoning,
            constitutional_influences=constitutional_influences,
            stakeholder_impacts=stakeholder_impacts,
            alternatives_considered=alternatives_considered,
            confidence_score=confidence_score
        )
    
    def generate_transparency_report(self, solution: DistributionSolution) -> Dict[str, Any]:
        """Generate comprehensive transparency report."""
        explanation = self.explain_solution(solution)
        
        return {
            "solution_summary": {
                "ethical_score": solution.ethical_score,
                "total_cost": solution.total_cost,
                "compliance_rate": np.mean(list(solution.compliance_status.values())),
                "overall_utility": solution.utilities.get("overall", 0)
            },
            "constitutional_compliance": solution.compliance_status,
            "stakeholder_utilities": {k: v for k, v in solution.utilities.items() if k != "overall"},
            "decision_explanation": explanation,
            "key_metrics": {
                "service_level": solution.metrics.get("service_level", 0),
                "carbon_emissions": solution.metrics.get("carbon_emissions_per_tonkm", 0),
                "underserved_service": solution.metrics.get("underserved_service_level", 0)
            },
            "transparency_score": explanation.confidence_score
        }

# Initialize transparency module
transparency = TransparencyModule(framework)

# Generate transparency report
print("🔍 Generating Transparency Report...")
transparency_report = transparency.generate_transparency_report(ethical_solution)

print(f"📊 Transparency Score: {transparency_report['transparency_score']:.3f}")
print(f"⚖️ Ethical Score: {transparency_report['solution_summary']['ethical_score']:.3f}")

# Display decision explanation
print("\n💡 Decision Explanation:")
for i, reason in enumerate(transparency_report['decision_explanation'].reasoning, 1):
    print(f"  {i}. {reason}")

print("\n📋 Constitutional Influences:")
for influence in transparency_report['decision_explanation'].constitutional_influences:
    print(f"  • {influence}")

print("\n👥 Stakeholder Impacts:")
for stakeholder, impact in transparency_report['decision_explanation'].stakeholder_impacts.items():
    print(f"  • {stakeholder}: {impact}")
```

## 6. Comparative Analysis with Previous Tiers

### Ethical Framework vs Traditional Optimization

Let's compare the ethical framework approach with traditional optimization methods to demonstrate the value added by constitutional constraints and multi-stakeholder consideration.

```python
# Generate traditional cost-optimized solution for comparison
def generate_traditional_solution(num_facilities: int, num_customers: int) -> DistributionSolution:
    """Generate traditional cost-optimized solution (without ethical constraints)."""
    
    # Traditional approach: minimize cost only
    facility_locations = [(5, 5), (15, 5), (25, 5), (35, 5)]  # Centralized for efficiency
    
    # Allocate based purely on distance (cost minimization)
    allocation_matrix = np.zeros((num_facilities, num_customers))
    customer_locations = [(np.random.uniform(0, 40), np.random.uniform(0, 10)) 
                          for _ in range(num_customers)]
    
    for customer_idx in range(num_customers):
        customer_loc = customer_locations[customer_idx]
        distances = [np.sqrt((x - customer_loc[0])**2 + (y - customer_loc[1])**2) 
                    for x, y in facility_locations]
        nearest_facility = np.argmin(distances)
        allocation_matrix[nearest_facility, customer_idx] = 1.0
    
    # Generate routes
    routes = []
    for facility_idx in range(num_facilities):
        served_customers = [i for i in range(num_customers) 
                          if allocation_matrix[facility_idx, i] > 0]
        if served_customers:
            routes.append([facility_idx] + served_customers)
    
    # Calculate metrics (traditional approach may have different values)
    traditional_metrics = {
        "service_level": 0.95,  # Higher service level (focus on profitable areas)
        "carbon_emissions_per_tonkm": 0.08,  # Higher emissions (cost optimization)
        "underserved_service_level": 0.65,  # Lower service for underserved
        "cost_overrun_percentage": 0.02,  # Lower cost (efficient)
        "resource_allocation_variance": 0.35,  # Higher variance (efficiency-focused)
        "delivery_time": 0.2,  # Faster delivery
        "reliability": 0.9,  # High reliability
        "job_security": 0.7,  # Lower job security (automation)
        "working_conditions": 0.7,  # Average conditions
        "fair_wages": 0.6,  # Lower wages (cost cutting)
        "local_employment": 0.5,  # Lower local employment
        "environmental_impact": 0.08,  # Higher environmental impact
        "infrastructure_development": 0.4,  # Less infrastructure development
        "resource_efficiency": 0.9,  # High resource efficiency
        "sustainability_score": 0.6,  # Lower sustainability
        "profitability": 0.9,  # High profitability
        "market_share": 0.8,  # High market share
        "risk_level": 0.3  # Moderate risk
    }
    
    # Calculate utilities
    traditional_utilities = ms_utility.calculate_overall_utility(traditional_metrics)
    
    # Check compliance (will likely fail some constraints)
    traditional_compliance = framework.check_compliance(traditional_metrics)
    
    # Create traditional solution
    traditional_solution = DistributionSolution(
        facility_locations=facility_locations,
        allocation_matrix=allocation_matrix,
        routes=routes,
        total_cost=35000,  # Lower cost
        metrics=traditional_metrics,
        utilities=traditional_utilities,
        compliance_status=traditional_compliance
    )
    
    traditional_solution.calculate_ethical_score(framework)
    return traditional_solution

# Generate traditional solution
print("⚖️ Generating Traditional Cost-Optimized Solution for Comparison...")
traditional_solution = generate_traditional_solution(4, 20)

# Compare solutions
print("\n📊 Solution Comparison:")
print(f"\n💰 Cost Comparison:")
print(f"  Ethical Framework: ${ethical_solution.total_cost:,.2f}")
print(f"  Traditional: ${traditional_solution.total_cost:,.2f}")
print(f"  Cost Difference: ${ethical_solution.total_cost - traditional_solution.total_cost:,.2f} ({((ethical_solution.total_cost / traditional_solution.total_cost - 1) * 100):.1f}% higher)")

print(f"\n⚖️ Ethical Score Comparison:")
print(f"  Ethical Framework: {ethical_solution.ethical_score:.3f}")
print(f"  Traditional: {traditional_solution.ethical_score:.3f}")
print(f"  Improvement: {(ethical_solution.ethical_score - traditional_solution.ethical_score):.3f}")

print(f"\n📋 Compliance Rate Comparison:")
ethical_compliance_rate = np.mean(list(ethical_solution.compliance_status.values()))
traditional_compliance_rate = np.mean(list(traditional_solution.compliance_status.values()))
print(f"  Ethical Framework: {ethical_compliance_rate:.1%}")
print(f"  Traditional: {traditional_compliance_rate:.1%}")
print(f"  Improvement: {(ethical_compliance_rate - traditional_compliance_rate) * 100:.1f} percentage points")

print(f"\n👥 Overall Utility Comparison:")
print(f"  Ethical Framework: {ethical_solution.utilities['overall']:.3f}")
print(f"  Traditional: {traditional_solution.utilities['overall']:.3f}")
print(f"  Difference: {(ethical_solution.utilities['overall'] - traditional_solution.utilities['overall']):.3f}")
```
