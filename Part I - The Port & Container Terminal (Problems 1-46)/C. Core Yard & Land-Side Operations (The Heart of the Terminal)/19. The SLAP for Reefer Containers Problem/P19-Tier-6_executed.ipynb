## Tier 6: Autonomous Self-Optimizing Ecosystem (Distributed Intelligence)

### Key Assumptions
- Multi-agent system with autonomous container and location agents
- Contract Net Protocol for distributed negotiation and coordination
- Market-based resource allocation with dynamic pricing mechanisms
- Self-organizing behavior through local interactions and emergence
- Learning and adaptation through experience and peer evaluation

### Approach (Step-by-Step)
1. **Agent Architecture**: Define autonomous agents with objectives and constraints
2. **Communication Protocols**: Implement Contract Net Protocol for auctions
3. **Market Mechanism**: Design dynamic pricing and bidding systems
4. **Coordination Logic**: Implement distributed decision making and consensus
5. **Learning Mechanism**: Enable agents to learn from experience
6. **Emergence Analysis**: Study system-wide emergent behaviors

### What to Look for in the Results
- Agent negotiation patterns and bidding strategies
- Market equilibrium and price convergence
- System performance vs centralized approaches
- Emergent behaviors and self-organization patterns
- Agent learning and adaptation over time
- Scalability and robustness analysis

```python
try:
    class AutonomousEcosystem:
        """Autonomous Self-Optimizing Ecosystem for Reefer Assignment"""
        
        def __init__(self, containers: List[ContainerAgent], 
                     locations: List[LocationAgent],
                     assignment_costs: Dict[Tuple[int, int], float]):
            self.containers = containers
            self.locations = locations
            self.assignment_costs = assignment_costs
            
            # Ecosystem components
            self.auctioneer = Auctioneer()
            self.marketplace = None  # Central marketplace
            self.message_bus = deque(maxlen=1000)
            self.contract_registry = {}  # Active contracts
            
            # Learning and adaptation
            self.global_performance = 0.0
            self.iteration = 0
            self.performance_history = []
            
            # Configuration
            self.max_auction_rounds = 3
            self.auction_timeout = 30  # seconds
            self.learning_enabled = True
            
            # Initialize marketplace
            self._initialize_marketplace()
            
        def _initialize_marketplace(self):
            """Initialize the central marketplace"""
            self.marketplace = {
                'active_auctions': [],
                'market_prices': {l.id: l.base_cost for l in self.locations},
                'market_demand': {c.id: 1.0 for c in self.containers},  # Initial demand
                'market_supply': {l.id: l.power_cap for l in self.locations},
                'transaction_history': []
            }
        
        def run_auction_round(self, round_num: int) -> Dict[str, Any]:
            """Run a single auction round"""
            print(f"\n🔬 Auction Round {round_num}:")
            
            # Initialize auction
            self.auctioneer.current_auction = {
                'round': round_num,
                'start_time': time.time(),
                'bids': [],
                'winning_bid': None,
                'contracts': [],
                'total_revenue': 0.0,
            }
            
            # Announce auction
            announcement = {
                'type': 'auction_announcement',
                'content': f"Starting Auction Round {round_num} for available locations"
            }
            self.message_bus.append(Message(
                sender="Auctioneer",
                receiver="All",
                message_type="announcement",
                content=announcement['content']
            ))
            
            # Container agents submit bids
            for container_agent in self.containers:
                if not container_agent.assigned:
                    bid = self._generate_bid(container_agent, round_num)
                    if bid:
                        self.auctioneer.current_auction['bids'].append(bid)
                        announcement['content'] += f"Container {container_agent.id} submitted bid: ${bid['price']:.2f}"
            
            # Location agents submit asks (reverse auction)
            for location_agent in self.locations:
                available_capacity = location_agent.power_cap - location_agent.power_usage
                if available_capacity > 0:
                    ask_price = self._generate_ask_price(location_agent, round_num)
                    if ask_price:
                        self.auctioneer.current_auction['bids'].append(ask_price)
                        announcement['content'] += f"Location {location_agent.id} submitted ask: ${ask_price['price']:.2f}"
            
            # Process auction
            self._process_auction()
            
            # Return auction results
            return {
                'round': round_num,
                'total_revenue': self.auctioneer.current_auction['total_revenue'],
                'contracts': self.auctioneer.current_auction['contracts'],
                'winning_bid': self.auctioneer.current_auction['winning_bid'],
                'participation': len(self.auctioneer.current_auction['bids']),
            }
        
        def _generate_bid(self, container_agent: ContainerAgent, round_num: int) -> Optional[Dict]:
            """Generate bid from container agent"""
            # Calculate container's valuation of current situation
            urgency_bonus = container_agent.urgency * 50  # Urgency premium
            time_pressure = container_agent.dwell_time / 120  # Time pressure factor
            
            # Find best available location
            best_location = None
            best_utility = 0.0
            
            for location_agent in self.locations:
                # Check if location can accommodate container
                if (location_agent.power_cap - location_agent.power_usage >= container_agent.energy and
                    location_agent.temp_min <= container_agent.temp_req <= location_agent.temp_max):
                    
                    # Calculate utility for this location
                    location_utilization = (location_agent.power_cap - location_agent.power_usage) / location_agent.power_cap
                    temp_compatibility = 1.0 if location_agent.temp_min <= container_agent.temp_req <= location_agent.temp_max else 0.0
                    
                    # Check container preferences
                    location_preference = container_agent.preferences.get(location_agent.location_type, 0.5)
                    
                    # Calculate total utility
                    utility = (location_utilization * 0.4 + 
                             temp_compatibility * 0.3 + 
                             location_preference * 0.3)
                    
                    if utility > best_utility:
                        best_utility = utility
                        best_location = location_agent
            
            if best_location:
                # Calculate bid price
                base_price = self.assignment_costs.get((container_agent.id, best_location.id), 100)
                
                # Adjust for container characteristics
                energy_factor = container_agent.energy / 50  # Normalize by typical energy
                urgency_factor = 1.0 + urgency_bonus
                time_factor = 1.0 + time_pressure
                flexibility_discount = 1.0 - (1.0 - container_agent.flexibility) * 0.2
                
                bid_price = base_price * energy_factor * urgency_factor * time_factor * flexibility_discount
                
                # Check budget constraint
                if bid_price <= container_agent.budget:
                    return {
                        'container_id': container_agent.id,
                        'location_id': best_location.id,
                        'price': bid_price,
                        'utility': best_utility,
                        'bid_type': 'bid'
                    }
            
            return None
        
        def _generate_ask_price(self, location_agent: LocationAgent, round_num: int) -> Optional[Dict]:
            """Generate ask price from location agent"""
            
            # Calculate location's valuation of current capacity
            capacity_utilization = location_agent.power_usage / location_agent.power_cap
            
            # Market demand for this location type
            market_demand = self.marketplace['market_demand'].get(location_agent.location_type, 1.0)
            
            # Location reputation
            reputation_factor = location_agent.reputation
            
            # Calculate ask price (inverse of utility)
            base_price = location_agent.base_cost
            
            # Adjust for market conditions
            market_factor = market_demand * reputation_factor
            capacity_premium = (1.0 + capacity_utilization) * 2.0  # Premium for high utilization
            
            ask_price = base_price * market_factor * capacity_premium
            
            return {
                'location_id': location_agent.id,
                'price': ask_price,
                'utility': 1.0 / ask_price if ask_price > 0 else 0,
                'ask_type': 'ask'
            }
        
        def _process_auction(self):
            """Process the auction using Contract Net Protocol"""
            
            auction = self.auctioneer.current_auction
            
            # Sort bids by price (descending for containers, ascending for locations)
            container_bids = [b for b in auction['bids'] if b['bid_type'] == 'bid']
            location_bids = [b for b in auction['bids'] if b['ask_type'] == 'ask']
            
            # Match highest bids with lowest asks
            matched_pairs = []
            
            while container_bids and location_bids:
                # Find best container bid
                best_bid = max(container_bids, key=lambda x: x['price'])
                # Find best location ask
                best_ask = min(location_bids, key=lambda x: x['price'])
                
                if best_bid['price'] >= best_ask['price']:
                    # Match found
                    matched_pairs.append((best_bid, best_ask))
                    container_bids.remove(best_bid)
                    location_bids.remove(best_ask)
                else:
                    # No match found, remove lowest bid
                    container_bids.remove(best_bid)
            
            # Create contracts for matched pairs
            for container_bid, location_ask in matched_pairs:
                price = (container_bid['price'] + location_ask['price']) / 2  # Average price
                duration = container_bid.get('duration', 24)  # Default 24 hours
                
                contract = Contract(
                    container_id=container_bid['container_id'],
                    location_id=location_ask['location_id'],
                    price=price,
                    duration=duration
                )
                auction['contracts'].append(contract)
                
                # Update agent states
                container_agent = next(c for c in self.containers if c.id == container_bid['container_id'])
                location_agent = next(l for l in self.locations if l.id == location_ask['location_id'])
                
                container_agent.assigned = True
                container_agent.current_location = location_agent.id
                container_agent.utility = container_bid['utility']
                container_agent.satisfaction += 0.1 * (container_agent.utility - 0.5)
                
                location_agent.assigned_containers.append(container_agent.id)
                location_agent.power_usage += container_agent.energy
                location_agent.utilization = location_agent.power_usage / location_agent.power_cap
                location_agent.revenue += price
                
                # Store contract
                contract_key = f"{container_bid['container_id']}_{location_ask['location_id']}"
                self.contract_registry[contract_key] = contract
                
            # Calculate auction metrics
            total_revenue = sum(contract.price for contract in auction['contracts'])
            auction['total_revenue'] = total_revenue
            
            # Update global performance
            if len(self.contract_registry) > 0:
                avg_price = total_revenue / len(auction['contracts'])
                self.global_performance = 100 / (avg_price / 20)  # Normalize by expected price
            
            auction['winning_bid'] = matched_pairs[0][0] if matched_pairs else None
            
            # Update learning
            self.iteration += 1
            self.performance_history.append({
                'iteration': self.iteration,
                'total_revenue': total_revenue,
                'contracts_created': len(auction['contracts']),
                'global_performance': self.global_performance
            })
        
        def analyze_emergent_behavior(self) -> Dict[str, Any]:
            """Analyze emergent behaviors in the ecosystem"""
            
            # Coalition formation patterns
            coalition_patterns = {}
            
            # Track coalition membership over time
            for contract_key, contract in self.contract_registry.items():
                # Extract cargo type from container ID
                container_id = contract_key.split('_')[0]
                location_id = contract_key.split('_')[1]
                
                # Find container and location agents
                container_agent = next(c for c in self.containers if c.id == int(container_id))
                location_agent = next(l for l in self.locations if l.id == int(location_id))
                
                # Determine coalition type based on cargo type
                if container_agent.cargo_type in ['Pharmaceutical', 'Vaccines']:
                    coalition_type = "Pharmaceutical Alliance"
                elif container_agent.cargo_type in ['Fresh Produce', 'Dairy', 'Beverages']:
                    coalition_type = "Fresh Produce Cooperative"
                else:
                    coalition_type = "Standard Freight Collective"
                
                coalition_key = f"{coalition_type}_{container_agent.cargo_type}"
                
                if coalition_key not in coalition_patterns:
                    coalition_patterns[coalition_key] = []
                coalition_patterns[coalition_key].append(location_id)
                
            # Pricing patterns
            pricing_patterns = {}
            for contract_key, contract in self.contract_registry.items():
                location_id = contract_key.split('_')[1]
                price = contract.price
                
                # Determine pricing strategy based on location type
                location_agent = next(l for l in self.locations if l.id == int(location_id))
                
                if location_agent.location_type in ['Ultra-Cold', 'Cold']:
                    pricing_pattern = "Premium Pricing"
                elif location_agent.location_type in ['General', 'Cool']:
                    pricing_pattern = "Standard Pricing"
                else:
                    pricing_pattern = "Economy Pricing"
                
                pricing_key = f"{pricing_pattern}_{location_agent.location_type}"
                
                if pricing_key not in pricing_patterns:
                    pricing_patterns[pricing_key] = []
                pricing_patterns[pricing_key].append(price)
                
            # System-wide metrics
            total_revenue = sum(contract.price for contract in self.contract_registry.values())
            avg_price = total_revenue / len(self.contract_registry)
            
            return {
                'coalition_patterns': coalition_patterns,
                'pricing_patterns': pricing_patterns,
                'total_revenue': total_revenue,
                'avg_price': avg_price,
                'contracts_created': len(self.contract_registry),
            }
        
        def get_system_state(self) -> Dict[str, Any]:
            """Get current system state"""
            
            # Agent states
            container_states = []
            for container in self.containers:
                container_states.append({
                    'id': container.id,
                    'assigned': container.assigned,
                    'location': container.current_location,
                    'utility': container.utility,
                    'satisfaction': container.satisfaction,
                    'urgency': container.urgency,
                    'flexibility': container.flexibility,
                    'budget': container.budget,
                })
            
            # Location states
            location_states = []
            for location in self.locations:
                location_states.append({
                    'id': location.id,
                    'assigned_containers': location.assigned_containers,
                    'power_usage': location.power_usage,
                    'utilization': location.utilization,
                    'revenue': location.revenue,
                    'market_price': location.market_price,
                    'reputation': location.reputation,
                    'location_type': location.location_type,
                })
            
            # Marketplace state
            marketplace_state = {
                'active_auctions': len(self.auctioneer.current_auction['bids']) if self.auctioneer.current_auction else 0,
                'total_contracts': len(self.contract_registry),
                'market_prices': self.marketplace['market_prices'],
                'total_revenue': sum(contract.price for contract in self.contract_registry.values()),
            }
            
            # Global performance
            global_performance = self.global_performance
            iteration = self.iteration
            performance_trend = self.performance_history[-10:] if len(self.performance_history) else []
            
            return {
                'container_states': container_states,
                'location_states': location_states,
                'marketplace_state': marketplace_state,
                'global_performance': global_performance,
                'iteration': iteration,
                'performance_trend': performance_trend,
            }
    
    print("Autonomous Ecosystem class defined")
except Exception as e:
    print(f'Error in cell: {e}')
```
