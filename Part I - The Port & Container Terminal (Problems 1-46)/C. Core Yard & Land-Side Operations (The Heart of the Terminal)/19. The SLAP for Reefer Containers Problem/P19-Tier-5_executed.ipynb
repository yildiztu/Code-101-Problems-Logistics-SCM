```python
@dataclass
class ReeferContainer:
    """Represents a reefer container with physical properties"""
    id: int
    energy: float  # Power consumption in kW
    temp_req: float  # Required temperature in Celsius
    temp_tolerance: float  # Temperature tolerance (±°C)
    dwell_time: float  # Dwell time in hours
    cargo_type: str  # Type of cargo
    cargo_value: float  # Value of cargo in thousands of dollars
    thermal_mass: float  # Thermal mass coefficient
    insulation_quality: float  # Insulation quality (0-1)

@dataclass
class StorageLocation:
    """Represents a storage location with physical infrastructure"""
    id: int
    power_cap: float  # Power capacity in kW
    temp_min: float  # Minimum temperature in Celsius
    temp_max: float  # Maximum temperature in Celsius
    base_cost: float  # Base cost multiplier
    backup_power: bool  # Has backup power capability
    sensor_count: int  # Number of IoT sensors
    location_type: str  # Type of storage (cold, cool, general)

@dataclass
class IoT_Sensor:
    """Represents an IoT sensor in the digital twin"""
    id: str
    sensor_type: str  # temperature, power, humidity, door, vibration
    location_id: int
    container_id: Optional[int]
    accuracy: float  # Sensor accuracy (0-1)
    sampling_rate: float  # Sampling rate in Hz
    battery_level: float  # Battery level (0-1)
    last_maintenance: datetime

print("Data classes defined for Digital Twin")
```

```python
class ReeferDigitalTwin:
    """Integrated Digital Twin for Reefer Container Terminal Operations"""
    
    def __init__(self, containers: List[ReeferContainer], 
                 locations: List[StorageLocation],
                 assignment_costs: Dict[Tuple[int, int], float]):
        self.containers = containers
        self.locations = locations
        self.assignment_costs = assignment_costs
        
        # Digital Twin state
        self.current_time = datetime.now()
        self.simulation_speed = 3600  # 1 second = 1 hour simulation
        
        # Physical layer state
        self.container_assignments = {}  # container_id -> location_id
        self.location_power_usage = {loc.id: 0.0 for loc in locations}
        self.container_temperatures = {c.id: c.temp_req for c in containers}
        self.power_grid_status = "normal"  # normal, warning, critical, failure
        
        # IoT sensors
        self.sensors = self._initialize_sensors()
        self.sensor_data = deque(maxlen=1000)  # Rolling sensor data
        
        # Data processing layer
        self.predictive_models = self._initialize_predictive_models()
        self.alerts = []
        self.kpi_history = deque(maxlen=100)
        
        # Application layer
        self.optimization_engine = None
        self.emergency_protocols = self._initialize_emergency_protocols()
        
        # Performance metrics
        self.total_cargo_value = sum(c.cargo_value for c in containers)
        self.cargo_integrity_preserved = 0.0
        self.energy_efficiency = 0.0
        
    def _initialize_sensors(self) -> List[IoT_Sensor]:
        """Initialize IoT sensors for the digital twin"""
        sensors = []
        sensor_id = 0
        
        # Location sensors (temperature, power, humidity)
        for location in self.locations:
            # Temperature sensors
            for i in range(3):  # 3 temperature sensors per location
                sensors.append(IoT_Sensor(
                    id=f"TEMP_L{location.id}_{i}",
                    sensor_type="temperature",
                    location_id=location.id,
                    container_id=None,
                    accuracy=0.95,
                    sampling_rate=0.1,  # Every 10 seconds
                    battery_level=1.0,
                    last_maintenance=datetime.now() - timedelta(days=30)
                ))
                sensor_id += 1
            
            # Power sensors
            for i in range(2):  # 2 power sensors per location
                sensors.append(IoT_Sensor(
                    id=f"POWER_L{location.id}_{i}",
                    sensor_type="power",
                    location_id=location.id,
                    container_id=None,
                    accuracy=0.98,
                    sampling_rate=1.0,  # Every 1 second
                    battery_level=1.0,
                    last_maintenance=datetime.now() - timedelta(days=15)
                ))
                sensor_id += 1
        
        # Container sensors (temperature, door, vibration)
        for container in self.containers:
            # Container temperature sensor
            sensors.append(IoT_Sensor(
                id=f"TEMP_C{container.id}",
                sensor_type="temperature",
                location_id=None,
                container_id=container.id,
                accuracy=0.97,
                sampling_rate=0.05,  # Every 20 seconds
                battery_level=0.85,
                last_maintenance=datetime.now() - timedelta(days=7)
            ))
            sensor_id += 1
            
            # Door sensor
            sensors.append(IoT_Sensor(
                id=f"DOOR_C{container.id}",
                sensor_type="door",
                location_id=None,
                container_id=container.id,
                accuracy=0.99,
                sampling_rate=10.0,  # Every 0.1 seconds
                battery_level=0.90,
                last_maintenance=datetime.now() - timedelta(days=5)
            ))
            sensor_id += 1
        
        return sensors
    
    def _initialize_predictive_models(self) -> Dict[str, Any]:
        """Initialize predictive models for the digital twin"""
        return {
            'temperature_failure': {
                'model_type': 'time_series',
                'accuracy': 0.947,
                'prediction_horizon': 24,  # hours
                'features': ['temp_trend', 'power_usage', 'door_openings', 'ambient_temp']
            },
            'power_failure': {
                'model_type': 'classification',
                'accuracy': 0.923,
                'prediction_horizon': 12,  # hours
                'features': ['power_trend', 'load_factor', 'weather_conditions', 'equipment_age']
            },
            'cargo_degradation': {
                'model_type': 'regression',
                'accuracy': 0.889,
                'prediction_horizon': 48,  # hours
                'features': ['temp_variance', 'power_stability', 'door_frequency', 'time_in_storage']
            }
        }
    
    def _initialize_emergency_protocols(self) -> Dict[str, Dict]:
        """Initialize emergency response protocols"""
        return {
            'power_failure': {
                'priority_containers': lambda c: c.cargo_type in ['Pharmaceutical', 'Vaccines'],
                'backup_power_locations': [l.id for l in self.locations if l.backup_power],
                'generator_capacity': 200,  # kW
                'response_time': 300  # seconds
            },
            'temperature_violation': {
                'critical_containers': lambda c: c.temp_tolerance <= 1.0,
                'relocation_priority': lambda c: c.cargo_value,
                'emergency_cooling': True,
                'response_time': 600  # seconds
            },
            'equipment_failure': {
                'affected_containers': [],
                'alternative_locations': [],
                'manual_override': True,
                'response_time': 900  # seconds
            }
        }
    
    def assign_containers(self, assignment: Dict[int, int]):
        """Assign containers to locations in the digital twin"""
        self.container_assignments = assignment
        
        # Update power usage
        self.location_power_usage = {loc.id: 0.0 for loc in self.locations}
        for container_id, location_id in assignment.items():
            container = next(c for c in self.containers if c.id == container_id)
            self.location_power_usage[location_id] += container.energy
        
        # Initialize container temperatures based on location
        for container_id, location_id in assignment.items():
            container = next(c for c in self.containers if c.id == container_id)
            location = next(l for l in self.locations if l.id == location_id)
            # Start at location's setpoint temperature
            self.container_temperatures[container_id] = (location.temp_min + location.temp_max) / 2
    
    def simulate_step(self, duration_hours: float = 1.0) -> Dict[str, Any]:
        """Simulate one time step in the digital twin"""
        step_results = {
            'timestamp': self.current_time,
            'duration': duration_hours,
            'sensor_readings': [],
            'alerts': [],
            'kpi': {},
            'predictions': {}
        }
        
        # Update time
        self.current_time += timedelta(hours=duration_hours)
        
        # Simulate sensor readings
        sensor_readings = self._simulate_sensor_readings()
        step_results['sensor_readings'] = sensor_readings
        
        # Store sensor data
        for reading in sensor_readings:
            self.sensor_data.append(reading)
        
        # Update physical state
        self._update_container_temperatures(duration_hours)
        self._update_power_grid_status()
        
        # Generate alerts
        alerts = self._generate_alerts()
        step_results['alerts'] = alerts
        self.alerts.extend(alerts)
        
        # Calculate KPIs
        kpi = self._calculate_kpi()
        step_results['kpi'] = kpi
        self.kpi_history.append(kpi)
        
        # Generate predictions
        predictions = self._generate_predictions()
        step_results['predictions'] = predictions
        
        return step_results
    
    def _simulate_sensor_readings(self) -> List[Dict[str, Any]]:
        """Simulate IoT sensor readings"""
        readings = []
        
        for sensor in self.sensors:
            reading = {
                'sensor_id': sensor.id,
                'sensor_type': sensor.sensor_type,
                'timestamp': self.current_time,
                'location_id': sensor.location_id,
                'container_id': sensor.container_id
            }
            
            # Generate reading based on sensor type
            if sensor.sensor_type == "temperature":
                if sensor.container_id:
                    # Container temperature sensor
                    actual_temp = self.container_temperatures[sensor.container_id]
                    noise = np.random.normal(0, 0.5)  # Sensor noise
                    reading['value'] = actual_temp + noise
                    reading['unit'] = "°C"
                else:
                    # Location temperature sensor
                    location = next(l for l in self.locations if l.id == sensor.location_id)
                    ambient_temp = (location.temp_min + location.temp_max) / 2
                    noise = np.random.normal(0, 0.3)
                    reading['value'] = ambient_temp + noise
                    reading['unit'] = "°C"
            
            elif sensor.sensor_type == "power":
                # Power sensor
                actual_power = self.location_power_usage[sensor.location_id]
                noise = np.random.normal(0, 1.0)
                reading['value'] = max(0, actual_power + noise)
                reading['unit'] = "kW"
            
            elif sensor.sensor_type == "door":
                # Door sensor (binary)
                reading['value'] = np.random.choice([0, 1], p=[0.95, 0.05])  # 5% door opening probability
                reading['unit'] = "binary"
            
            # Add sensor health information
            reading['battery_level'] = sensor.battery_level
            reading['accuracy'] = sensor.accuracy
            reading['signal_strength'] = np.random.uniform(0.7, 1.0)
            
            readings.append(reading)
        
        return readings
    
    def _update_container_temperatures(self, duration_hours: float):
        """Update container temperatures based on physics simulation"""
        for container_id, location_id in self.container_assignments.items():
            container = next(c for c in self.containers if c.id == container_id)
            location = next(l for l in self.locations if l.id == location_id)
            
            current_temp = self.container_temperatures[container_id]
            target_temp = (location.temp_min + location.temp_max) / 2
            
            # Simple thermal dynamics model
            # dT/dt = k * (T_ambient - T_container)
            thermal_coefficient = 0.1 * container.thermal_mass * container.insulation_quality
            temp_change = thermal_coefficient * (target_temp - current_temp) * duration_hours
            
            # Add random fluctuations
            fluctuation = np.random.normal(0, 0.1) * np.sqrt(duration_hours)
            
            new_temp = current_temp + temp_change + fluctuation
            self.container_temperatures[container_id] = new_temp
    
    def _update_power_grid_status(self):
        """Update power grid status based on load and conditions"""
        total_load = sum(self.location_power_usage.values())
        total_capacity = sum(loc.power_cap for loc in self.locations)
        load_factor = total_load / total_capacity
        
        # Add weather conditions (simplified)
        weather_factor = np.random.uniform(0.9, 1.1)  # Weather impact on grid
        
        effective_load_factor = load_factor * weather_factor
        
        if effective_load_factor < 0.7:
            self.power_grid_status = "normal"
        elif effective_load_factor < 0.85:
            self.power_grid_status = "warning"
        elif effective_load_factor < 0.95:
            self.power_grid_status = "critical"
        else:
            self.power_grid_status = "failure"
    
    def _generate_alerts(self) -> List[Dict[str, Any]]:
        """Generate alerts based on sensor data and system state"""
        alerts = []
        
        # Temperature violation alerts
        for container_id, location_id in self.container_assignments.items():
            container = next(c for c in self.containers if c.id == container_id)
            location = next(l for l in self.locations if l.id == location_id)
            current_temp = self.container_temperatures[container_id]
            
            # Check temperature tolerance
            if abs(current_temp - container.temp_req) > container.temp_tolerance:
                alerts.append({
                    'type': 'temperature_violation',
                    'severity': 'high' if abs(current_temp - container.temp_req) > container.temp_tolerance * 2 else 'medium',
                    'container_id': container_id,
                    'location_id': location_id,
                    'current_temp': current_temp,
                    'required_temp': container.temp_req,
                    'deviation': abs(current_temp - container.temp_req),
                    'timestamp': self.current_time,
                    'message': f"Container {container_id} temperature deviation: {current_temp:.1f}°C (required: {container.temp_req}°C)"
                })
        
        # Power overload alerts
        for location in self.locations:
            usage = self.location_power_usage[location.id]
            if usage > location.power_cap:
                alerts.append({
                    'type': 'power_overload',
                    'severity': 'critical' if usage > location.power_cap * 1.2 else 'high',
                    'location_id': location.id,
                    'current_usage': usage,
                    'capacity': location.power_cap,
                    'overload': usage - location.power_cap,
                    'timestamp': self.current_time,
                    'message': f"Location {location.id} power overload: {usage:.1f}kW (capacity: {location.power_cap}kW)"
                })
        
        # Grid status alerts
        if self.power_grid_status in ["critical", "failure"]:
            alerts.append({
                'type': 'power_grid',
                'severity': 'critical' if self.power_grid_status == "failure" else 'high',
                'status': self.power_grid_status,
                'timestamp': self.current_time,
                'message': f"Power grid status: {self.power_grid_status}"
            })
        
        return alerts
    
    def _calculate_kpi(self) -> Dict[str, float]:
        """Calculate Key Performance Indicators"""
        # Cargo integrity preservation
        containers_in_tolerance = 0
        for container_id, location_id in self.container_assignments.items():
            container = next(c for c in self.containers if c.id == container_id)
            current_temp = self.container_temperatures[container_id]
            if abs(current_temp - container.temp_req) <= container.temp_tolerance:
                containers_in_tolerance += 1
        
        self.cargo_integrity_preserved = (containers_in_tolerance / len(self.container_assignments)) * 100 if self.container_assignments else 0
        
        # Energy efficiency
        total_energy_consumed = sum(self.location_power_usage.values())
        total_energy_required = sum(c.energy for c in self.containers if c.id in self.container_assignments)
        self.energy_efficiency = (total_energy_required / total_energy_consumed) * 100 if total_energy_consumed > 0 else 0
        
        # Location utilization
        total_utilization = 0
        for location in self.locations:
            utilization = (self.location_power_usage[location.id] / location.power_cap) * 100
            total_utilization += utilization
        avg_utilization = total_utilization / len(self.locations)
        
        # Alert rate
        recent_alerts = [a for a in self.alerts if a['timestamp'] > self.current_time - timedelta(hours=1)]
        alert_rate = len(recent_alerts)
        
        return {
            'cargo_integrity': self.cargo_integrity_preserved,
            'energy_efficiency': self.energy_efficiency,
            'avg_utilization': avg_utilization,
            'alert_rate': alert_rate,
            'containers_assigned': len(self.container_assignments),
            'total_containers': len(self.containers)
        }
    
    def _generate_predictions(self) -> Dict[str, Any]:
        """Generate predictive insights"""
        predictions = {}
        
        # Temperature failure predictions
        temp_violation_risk = 0.0
        for container_id, location_id in self.container_assignments.items():
            container = next(c for c in self.containers if c.id == container_id)
            current_temp = self.container_temperatures[container_id]
            
            # Simplified risk calculation
            temp_deviation = abs(current_temp - container.temp_req)
            risk_factor = min(1.0, temp_deviation / container.temp_tolerance)
            temp_violation_risk += risk_factor
        
        temp_violation_risk = (temp_violation_risk / len(self.container_assignments)) * 100 if self.container_assignments else 0
        
        predictions['temperature_failure_risk'] = temp_violation_risk
        
        # Power failure risk
        total_load = sum(self.location_power_usage.values())
        total_capacity = sum(loc.power_cap for loc in self.locations)
        load_factor = total_load / total_capacity
        
        power_failure_risk = max(0, (load_factor - 0.8) * 100)  # Risk increases above 80% load
        predictions['power_failure_risk'] = power_failure_risk
        
        # Cargo degradation risk
        degradation_risk = 0.0
        for container_id, location_id in self.container_assignments.items():
            container = next(c for c in self.containers if c.id == container_id)
            current_temp = self.container_temperatures[container_id]
            
            # Risk based on temperature stability and cargo value
            temp_stability = 1.0 - min(1.0, abs(current_temp - container.temp_req) / (container.temp_tolerance * 2))
            value_factor = container.cargo_value / 1000  # Normalize by max value
            degradation_risk += (1.0 - temp_stability) * value_factor
        
        degradation_risk = (degradation_risk / len(self.container_assignments)) * 100 if self.container_assignments else 0
        predictions['cargo_degradation_risk'] = degradation_risk
        
        return predictions
    
    def run_scenario(self, scenario_name: str, duration_hours: float = 24.0) -> Dict[str, Any]:
        """Run a specific scenario simulation"""
        scenario_results = {
            'scenario': scenario_name,
            'duration': duration_hours,
            'steps': [],
            'summary': {},
            'alerts': [],
            'kpi_trend': []
        }
        
        if scenario_name == "normal_operation":
            # Normal operation scenario
            for hour in range(int(duration_hours)):
                step_result = self.simulate_step(1.0)
                scenario_results['steps'].append(step_result)
                scenario_results['kpi_trend'].append(step_result['kpi'])
        
        elif scenario_name == "hurricane_patricia":
            # Hurricane Patricia scenario
            for hour in range(int(duration_hours)):
                # Simulate hurricane conditions
                if hour == 6:  # Hurricane hits at hour 6
                    # Power grid failure
                    self.power_grid_status = "failure"
                    # Reduce power capacity
                    for location in self.locations:
                        if not location.backup_power:
                            self.location_power_usage[location.id] = location.power_cap * 0.3  # Emergency power only
                
                step_result = self.simulate_step(1.0)
                
                # Emergency response at hour 7
                if hour == 7:
                    emergency_actions = self._execute_emergency_protocols()
                    step_result['emergency_actions'] = emergency_actions
                
                scenario_results['steps'].append(step_result)
                scenario_results['kpi_trend'].append(step_result['kpi'])
                scenario_results['alerts'].extend(step_result['alerts'])
        
        elif scenario_name == "sensor_failure":
            # Sensor failure scenario
            for hour in range(int(duration_hours)):
                # Simulate sensor degradation
                if hour == 8:
                    # 20% of sensors fail
                    failed_sensors = random.sample(self.sensors, int(len(self.sensors) * 0.2))
                    for sensor in failed_sensors:
                        sensor.accuracy *= 0.5  # Reduced accuracy
                        sensor.battery_level *= 0.7  # Battery drain
                
                step_result = self.simulate_step(1.0)
                scenario_results['steps'].append(step_result)
                scenario_results['kpi_trend'].append(step_result['kpi'])
        
        # Generate scenario summary
        if scenario_results['kpi_trend']:
            final_kpi = scenario_results['kpi_trend'][-1]
            initial_kpi = scenario_results['kpi_trend'][0]
            
            scenario_results['summary'] = {
                'final_cargo_integrity': final_kpi['cargo_integrity'],
                'initial_cargo_integrity': initial_kpi['cargo_integrity'],
                'integrity_preservation': final_kpi['cargo_integrity'],
                'final_energy_efficiency': final_kpi['energy_efficiency'],
                'total_alerts': len(scenario_results['alerts']),
                'high_severity_alerts': len([a for a in scenario_results['alerts'] if a['severity'] == 'critical' or a['severity'] == 'high'])
            }
        
        return scenario_results
    
    def _execute_emergency_protocols(self) -> Dict[str, Any]:
        """Execute emergency response protocols"""
        actions = {
            'timestamp': self.current_time,
            'protocols_executed': [],
            'containers_moved': [],
            'backup_power_activated': [],
            'critical_containers_protected': []
        }
        
        # Power failure protocol
        if self.power_grid_status == "failure":
            protocol = self.emergency_protocols['power_failure']
            actions['protocols_executed'].append('power_failure')
            
            # Identify priority containers
            priority_containers = [c for c in self.containers if protocol['priority_containers'](c)]
            backup_locations = protocol['backup_power_locations']
            
            # Move priority containers to backup power locations
            moved_count = 0
            for container in priority_containers[:23]:  # Move 23 priority containers
                if container.id in self.container_assignments:
                    current_location = self.container_assignments[container.id]
                    if current_location not in backup_locations:
                        # Find available backup location
                        for backup_loc_id in backup_locations:
                            backup_loc = next(l for l in self.locations if l.id == backup_loc_id)
                            if (self.location_power_usage[backup_loc_id] + container.energy <= backup_loc.power_cap and
                                backup_loc.temp_min <= container.temp_req <= backup_loc.temp_max):
                                # Move container
                                self.container_assignments[container.id] = backup_loc_id
                                self.location_power_usage[current_location] -= container.energy
                                self.location_power_usage[backup_loc_id] += container.energy
                                actions['containers_moved'].append({
                                    'container_id': container.id,
                                    'from_location': current_location,
                                    'to_location': backup_loc_id,
                                    'reason': 'priority_backup_power'
                                })
                                moved_count += 1
                                break
            
            actions['critical_containers_protected'] = moved_count
            actions['backup_power_activated'] = backup_locations
        
        return actions

print("Digital Twin class defined")
```
