# 16. The Storage Location Assignment Problem (SLAP)
## Tier 6: The Autonomous & Self-Optimizing Ecosystem - Distributed Intelligence

### Key assumptions
- Storage locations operate as autonomous intelligent agents
- Products are represented as agent entities with their own objectives
- Multi-agent coordination occurs through distributed market mechanisms
- Game-theoretic negotiation protocols drive emergent system-wide optimization

### Approach (step-by-step)
1. **Agent Architecture**: Define location agents with local intelligence capabilities
2. **Product Agents**: Create SKU agents with multi-criteria utility functions
3. **Market Mechanism**: Implement distributed auction and negotiation protocols
4. **Bidding Strategy**: Agents bid for high-value products using utility maximization
5. **Contract Formation**: Establish bilateral agreements with reputation scoring
6. **Emergent Behavior**: Observe self-organization and system-wide optimization

```python
from typing import Tuple, List, Dict, Set
from itertools import product

# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass
from typing import List, Tuple, Dict, Optional, Any
import random
import time
from datetime import datetime, timedelta
from collections import defaultdict, deque
from enum import Enum
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
class AgentState(Enum):
    """Agent states in the ecosystem"""
    ACTIVE = "active"
    NEGOTIATING = "negotiating"
    CONTRACTED = "contracted"
    IDLE = "idle"
    MAINTENANCE = "maintenance"

@dataclass
class AgentCapability:
    """Defines agent capabilities"""
    capacity: float
    speed: float
    reliability: float
    cost_per_unit: float
    specialization: List[str]

@dataclass
class UtilityFunction:
    """Multi-criteria utility function for agents"""
    weight_cost: float = 0.3
    weight_speed: float = 0.25
    weight_reliability: float = 0.2
    weight_proximity: float = 0.15
    weight_reputation: float = 0.1

@dataclass
class Contract:
    """Represents a contract between agents"""
    contract_id: str
    buyer_id: str
    seller_id: str
    product_id: str
    terms: Dict[str, Any]
    duration: timedelta
    penalty_rate: float
    created_at: datetime
    status: str = "active"

@dataclass
class Bid:
    """Represents a bid in the auction market"""
    bid_id: str
    bidder_id: str
    product_id: str
    price: float
    utility_score: float
    terms: Dict[str, Any]
    timestamp: datetime
```

```python
try:
    class LocationAgent:
        """Autonomous storage location agent"""
    
        def __init__(self, agent_id: str, location_type: str, position: Tuple[float, float],
                     capability: AgentCapability, utility_function: UtilityFunction):
            self.agent_id = agent_id
            self.location_type = location_type  # 'premium', 'standard', 'bulk'
            self.position = position
            self.capability = capability
            self.utility_function = utility_function
    
            # Agent state
            self.state = AgentState.ACTIVE
            self.current_load = 0.0
           .contracts = []
            self.reputation_score = 0.5  # Starts neutral
    
            # Market participation
           .available_capacity = capability.capacity
           .asking_price = capability.cost_per_unit
    
            # Learning and adaptation
           .success_history = deque(maxlen=100)
           .profit_history = deque(maxlen=100)
    
        def calculate_utility(self, product_agent: 'ProductAgent', bid_price: float) -> float:
            """Calculate utility for hosting a product agent"""
    
            # Cost utility (lower cost = higher utility)
            cost_utility = 1.0 - (bid_price / (self.capability.cost_per_unit * 2))
            cost_utility = max(0, min(1, cost_utility))
    
            # Speed utility (faster processing = higher utility)
            speed_utility = self.capability.speed / 10.0  # Normalize to 0-1
    
            # Reliability utility
            reliability_utility = self.capability.reliability
    
            # Proximity utility (distance to I/O point)
            distance_to_io = np.sqrt(self.position[0]**2 + self.position[1]**2)
            proximity_utility = 1.0 - (distance_to_io / 1000.0)  # Normalize to 0-1
            proximity_utility = max(0, min(1, proximity_utility))
    
            # Reputation utility
            reputation_utility = self.reputation_score
    
            # Weighted combination
            total_utility = (
                self.utility_function.weight_cost * cost_utility +
                self.utility_function.weight_speed * speed_utility +
                self.utility_function.weight_reliability * reliability_utility +
                self.utility_function.weight_proximity * proximity_utility +
                self.utility_function.weight_reputation * reputation_utility
            )
    
            return total_utility
    
        def submit_bid(self, product_agent: 'ProductAgent', auction_id: str) -> Bid:
            """Submit a bid for a product agent"""
    
            # Calculate optimal bid price
            base_price = self.capability.cost_per_unit
    
            # Adjust based on demand and current load
            load_factor = self.current_load / self.capability.capacity
            demand_multiplier = 1.0 + (load_factor * 0.5)  # Higher price when loaded
    
            # Adjust based on product characteristics
            if product_agent.is_high_velocity:
                demand_multiplier *= 1.2  # Premium for high-velocity products
    
            if product_agent.product_category in self.capability.specialization:
                demand_multiplier *= 0.9  # Discount for specialized products
    
            bid_price = base_price * demand_multiplier
    
            # Calculate utility score
            utility_score = self.calculate_utility(product_agent, bid_price)
    
            # Create bid
            bid = Bid(
                bid_id=f"{self.agent_id}_{auction_id}_{int(time.time())}",
                bidder_id=self.agent_id,
                product_id=product_agent.product_id,
                price=bid_price,
                utility_score=utility_score,
                terms={
                    "service_level": self.capability.reliability,
                    "processing_speed": self.capability.speed,
                    "specialization": self.capability.specialization
                },
                timestamp=datetime.now()
            )
    
            return bid
    
        def execute_contract(self, contract: Contract):
            """Execute a contract with a product agent"""
    
            self.contracts.append(contract)
            self.current_load += 1.0  # Simplified load calculation
            self.available_capacity = self.capability.capacity - self.current_load
    
            # Update state
            self.state = AgentState.CONTRACTED
    
        def update_reputation(self, success: bool, profit: float):
            """Update reputation based on contract performance"""
    
            self.success_history.append(success)
            self.profit_history.append(profit)
    
            # Calculate reputation update
            if len(self.success_history) >= 10:
                recent_success_rate = sum(self.success_history[-10:]) / 10
    
                # Reputation moves toward recent performance
                self.reputation_score = 0.7 * self.reputation_score + 0.3 * recent_success_rate
                self.reputation_score = max(0, min(1, self.reputation_score))
    
    class ProductAgent:
        """Autonomous product/SKU agent"""
    
        def __init__(self, product_id: str, velocity: float, size: float,
                     category: str, is_high_velocity: bool = False):
            self.product_id = product_id
            self.velocity = velocity  # Demand frequency
            self.size = size
            self.product_category = category
            self.is_high_velocity = is_high_velocity
    
            # Agent state
            self.current_location = None
            self.contracts = []
            self.utility_function = UtilityFunction()
    
            # Requirements and preferences
           .min_reliability = 0.8
            .max_cost = velocity * 10.0  # Willing to pay up to 10x velocity
            .preferred_specialization = [category]
    
        def evaluate_bids(self, bids: List[Bid]) -> Tuple[Bid, float]:
            """Evaluate bids and select the best one"""
    
            if not bids:
                return None, 0.0
    
            best_bid = None
            best_score = -float('inf')
    
            for bid in bids:
                # Check basic constraints
                if bid.price > self.max_cost:
                    continue
    
                if bid.terms["service_level"] < self.min_reliability:
                    continue
    
                # Calculate product utility score
                cost_score = 1.0 - (bid.price / self.max_cost)
                speed_score = bid.terms["processing_speed"] / 10.0
                reliability_score = bid.terms["service_level"]
    
                # Specialization bonus
                specialization_bonus = 0.0
                if self.product_category in bid.terms["specialization"]:
                    specialization_bonus = 0.2
    
                total_score = (
                    0.4 * cost_score +
                    0.3 * speed_score +
                    0.2 * reliability_score +
                    0.1 * specialization_bonus
                )
    
                if total_score > best_score:
                    best_score = total_score
                    best_bid = bid
    
            return best_bid, best_score
    
        def accept_contract(self, bid: Bid, location_agent: LocationAgent):
            """Accept a contract and move to the location"""
    
            # Create contract
            contract = Contract(
                contract_id=f"CONTRACT_{self.product_id}_{location_agent.agent_id}_{int(time.time())}",
                buyer_id=self.product_id,
                seller_id=location_agent.agent_id,
                product_id=self.product_id,
                terms={
                    "price": bid.price,
                    "service_level": bid.terms["service_level"],
                    "processing_speed": bid.terms["processing_speed"]
                },
                duration=timedelta(days=30),  # 30-day contract
                penalty_rate=0.1,
                created_at=datetime.now()
            )
    
            self.contracts.append(contract)
            self.current_location = location_agent.agent_id
    
            # Execute contract on location agent
            location_agent.execute_contract(contract)
    
            return contract
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    class DistributedMarket:
        """Distributed auction market for agent coordination"""
    
        def __init__(self, market_id: str):
            self.market_id = market_id
           .active_auctions = {}
           .market_history = []
           .price_discovery = defaultdict(list)
    
        def create_auction(self, product_agent: ProductAgent, duration_minutes: int = 5) -> str:
            """Create a new auction for a product agent"""
    
            auction_id = f"AUCTION_{product_agent.product_id}_{int(time.time())}"
    
            auction = {
                "auction_id": auction_id,
                "product_agent": product_agent,
                "bids": [],
                "start_time": datetime.now(),
                "end_time": datetime.now() + timedelta(minutes=duration_minutes),
                "status": "active",
                "winner": None
            }
    
            self.active_auctions[auction_id] = auction
            return auction_id
    
        def submit_bid(self, auction_id: str, bid: Bid):
            """Submit a bid to an active auction"""
    
            if auction_id in self.active_auctions:
                auction = self.active_auctions[auction_id]
    
                if auction["status"] == "active" and datetime.now() < auction["end_time"]:
                    auction["bids"].append(bid)
                    return True
    
            return False
    
        def close_auction(self, auction_id: str) -> Optional[Contract]:
            """Close an auction and determine the winner"""
    
            if auction_id not in self.active_auctions:
                return None
    
            auction = self.active_auctions[auction_id]
            auction["status"] = "closed"
    
            # Evaluate bids
            product_agent = auction["product_agent"]
            best_bid, best_score = product_agent.evaluate_bids(auction["bids"])
    
            if best_bid:
                auction["winner"] = best_bid.bidder_id
    
                # Record price discovery
                self.price_discovery[product_agent.product_category].append(best_bid.price)
    
                # Record market history
                self.market_history.append({
                    "auction_id": auction_id,
                    "product_id": product_agent.product_id,
                    "category": product_agent.product_category,
                    "winning_bid": best_bid.price,
                    "num_bids": len(auction["bids"]),
                    "timestamp": datetime.now()
                })
    
            # Remove from active auctions
            del self.active_auctions[auction_id]
    
            return best_bid
    
        def get_market_stats(self) -> Dict[str, Any]:
            """Get market statistics and price trends"""
    
            if not self.market_history:
                return {"status": "no_data"}
    
            # Calculate statistics
            total_auctions = len(self.market_history)
            avg_bids_per_auction = np.mean([h["num_bids"] for h in self.market_history])
    
            # Price trends by category
            price_trends = {}
            for category, prices in self.price_discovery.items():
                if len(prices) >= 2:
                    price_trends[category] = {
                        "current_price": prices[-1],
                        "avg_price": np.mean(prices),
                        "price_volatility": np.std(prices),
                        "trend": "up" if prices[-1] > prices[-2] else "down"
                    }
    
            return {
                "total_auctions": total_auctions,
                "active_auctions": len(self.active_auctions),
                "avg_bids_per_auction": avg_bids_per_auction,
                "price_trends": price_trends,
                "market_efficiency": min(1.0, avg_bids_per_auction / 5.0)  # Efficiency metric
            }
    
    class AutonomousEcosystem:
        """Autonomous self-optimizing ecosystem for SLAP"""
    
        def __init__(self, n_locations: int = 500, n_products: int = 100000):
            self.n_locations = n_locations
            self.n_products = n_products
    
            # Initialize agents
            self.location_agents = []
            self.product_agents = []
    
            # Market mechanism
            self.market = DistributedMarket("SLAP_MARKET_001")
    
            # Ecosystem state
            self.current_time = datetime.now()
            self.simulation_running = False
    
            # Performance tracking
            self.performance_history = []
            self.emergent_behaviors = []
    
            self._initialize_ecosystem()
    
        def _initialize_ecosystem(self):
            """Initialize the autonomous agent ecosystem"""
    
            print(f"Initializing Autonomous Ecosystem:")
            print(f"  Location agents: {self.n_locations:,}")
            print(f"  Product agents: {self.n_products:,}")
    
            # Create location agents
            location_types = ['premium', 'standard', 'bulk']
            type_distribution = [0.1, 0.3, 0.6]  # 10% premium, 30% standard, 60% bulk
    
            for i in range(self.n_locations):
                # Determine location type
                location_type = np.random.choice(location_types, p=type_distribution)
    
                # Position (simplified layout)
                position = (
                    np.random.uniform(0, 1000),  # x coordinate
                    np.random.uniform(0, 500)    # y coordinate
                )
    
                # Capability based on type
                if location_type == 'premium':
                    capability = AgentCapability(
                        capacity=100.0,
                        speed=9.0,
                        reliability=0.98,
                        cost_per_unit=15.0,
                        specialization=['electronics', 'pharma', 'luxury']
                    )
                elif location_type == 'standard':
                    capability = AgentCapability(
                        capacity=200.0,
                        speed=6.0,
                        reliability=0.92,
                        cost_per_unit=8.0,
                        specialization=['clothing', 'books', 'home']
                    )
                else:  # bulk
                    capability = AgentCapability(
                        capacity=500.0,
                        speed=3.0,
                        reliability=0.85,
                        cost_per_unit=3.0,
                        specialization=['furniture', 'sports', 'seasonal']
                    )
    
                # Utility function (varies by agent type)
                utility_function = UtilityFunction(
                    weight_cost=np.random.uniform(0.2, 0.4),
                    weight_speed=np.random.uniform(0.2, 0.3),
                    weight_reliability=np.random.uniform(0.15, 0.25),
                    weight_proximity=np.random.uniform(0.1, 0.2),
                    weight_reputation=np.random.uniform(0.05, 0.15)
                )
    
                agent = LocationAgent(
                    agent_id=f"LOC_{i:04d}",
                    location_type=location_type,
                    position=position,
                    capability=capability,
                    utility_function=utility_function
                )
    
                self.location_agents.append(agent)
    
            # Create product agents
            product_categories = [
                'electronics', 'pharma', 'clothing', 'books', 'home',
                'furniture', 'sports', 'seasonal', 'luxury', 'toys'
            ]
    
            for i in range(self.n_products):
                category = np.random.choice(product_categories)
    
                # Velocity distribution (few high-velocity, many low-velocity)
                if np.random.random() < 0.05:  # 5% high-velocity
                    velocity = np.random.uniform(80, 100)
                    is_high_velocity = True
                elif np.random.random() < 0.15:  # 15% medium-velocity
                    velocity = np.random.uniform(40, 80)
                    is_high_velocity = False
                else:  # 80% low-velocity
                    velocity = np.random.uniform(5, 40)
                    is_high_velocity = False
    
                size = np.random.uniform(0.5, 5.0)  # Size units
    
                agent = ProductAgent(
                    product_id=f"SKU_{i:06d}",
                    velocity=velocity,
                    size=size,
                    category=category,
                    is_high_velocity=is_high_velocity
                )
    
                self.product_agents.append(agent)
    
            print(f"Ecosystem initialized successfully!")
            print(f"  High-velocity products: {sum(1 for p in self.product_agents if p.is_high_velocity):,}")
            print(f"  Premium locations: {len([l for l in self.location_agents if l.location_type == 'premium']):,}")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
        def run_simulation(self, duration_minutes: int = 30, demand_scenario: str = "normal"):
            """Run the autonomous ecosystem simulation"""
    
            print(f"\nStarting Autonomous Ecosystem Simulation")
            print(f"Duration: {duration_minutes} minutes")
            print(f"Demand Scenario: {demand_scenario}")
    
            self.simulation_running = True
            start_time = self.current_time
            end_time = start_time + timedelta(minutes=duration_minutes)
    
            # Apply demand scenario
            self._apply_demand_scenario(demand_scenario)
    
            auction_cycle = 0
    
            while self.current_time < end_time and self.simulation_running:
                # Run auction cycle
                contracts_formed = self._run_auction_cycle()
                auction_cycle += 1
    
                # Update agent reputations and learning
                self._update_agent_learning()
    
                # Detect emergent behaviors
                emergent_behavior = self._detect_emergent_behaviors()
                if emergent_behavior:
                    self.emergent_behaviors.append(emergent_behavior)
    
                # Record performance
                performance = self._calculate_performance()
                self.performance_history.append({
                    "timestamp": self.current_time,
                    "auction_cycle": auction_cycle,
                    "contracts_formed": contracts_formed,
                    "performance": performance
                })
    
                # Progress reporting
                if auction_cycle % 10 == 0:
                    elapsed = (self.current_time - start_time).total_seconds() / 60
                    print(f"[{self.current_time.strftime('%H:%M:%S')}] Cycle {auction_cycle}: {elapsed:.0f}/{duration_minutes} min")
                    print(f"  Contracts formed: {contracts_formed:,}")
                    print(f"  System performance: {performance['efficiency']:.3f}")
                    if emergent_behavior:
                        print(f"  Emergent behavior: {emergent_behavior['type']}")
    
                # Advance time
                self.current_time += timedelta(seconds=30)
    
            print(f"\nSimulation completed! Total cycles: {auction_cycle}")
            self.simulation_running = False
    
        def _apply_demand_scenario(self, scenario: str):
            """Apply demand scenario to product agents"""
    
            if scenario == "black_friday":
                print("Applying Black Friday demand surge...")
    
                # Increase velocity for electronics and luxury products
                for product in self.product_agents:
                    if product.product_category in ['electronics', 'luxury']:
                        product.velocity *= 3.5  # 250% increase
                        product.is_high_velocity = True
                        product.max_cost = product.velocity * 12.0  # Willing to pay more
                    elif product.product_category in ['clothing', 'sports']:
                        product.velocity *= 2.0  # 100% increase
    
            elif scenario == "flu_season":
                print("Applying flu season demand pattern...")
    
                # Increase velocity for pharma and healthcare products
                for product in self.product_agents:
                    if product.product_category in ['pharma', 'healthcare']:
                        product.velocity *= 4.0  # 300% increase
                        product.is_high_velocity = True
    
        def _run_auction_cycle(self) -> int:
            """Run one cycle of auctions"""
    
            contracts_formed = 0
    
            # Select products needing contracts (unassigned or expiring)
            products_needing_contracts = [
                p for p in self.product_agents
                if p.current_location is None or len(p.contracts) == 0
            ]
    
            # Limit to manageable number per cycle
            max_auctions_per_cycle = min(1000, len(products_needing_contracts))
            selected_products = np.random.choice(
                products_needing_contracts, max_auctions_per_cycle, replace=False
            )
    
            # Create and run auctions
            for product in selected_products:
                # Create auction
                auction_id = self.market.create_auction(product, duration_minutes=1)
    
                # Collect bids from location agents
                available_locations = [
                    l for l in self.location_agents
                    if l.available_capacity > 0 and l.state == AgentState.ACTIVE
                ]
    
                # Limit bidding to nearby locations for efficiency
                max_bidders = min(50, len(available_locations))
                bidding_locations = np.random.choice(available_locations, max_bidders, replace=False)
    
                for location in bidding_locations:
                    bid = location.submit_bid(product, auction_id)
                    self.market.submit_bid(auction_id, bid)
    
                # Close auction and form contract
                winning_bid = self.market.close_auction(auction_id)
    
                if winning_bid:
                    # Find the winning location agent
                    winning_location = next(l for l in self.location_agents if l.agent_id == winning_bid.bidder_id)
    
                    # Form contract
                    contract = product.accept_contract(winning_bid, winning_location)
                    contracts_formed += 1
    
            return contracts_formed
    
        def _update_agent_learning(self):
            """Update agent learning and reputation"""
    
            # Update location agent reputations based on contract performance
            for location in self.location_agents:
                if location.contracts:
                    # Simulate contract performance
                    for contract in location.contracts[-5:]:  # Last 5 contracts
                        success = np.random.random() < location.capability.reliability
                        profit = contract.terms["price"] * np.random.uniform(0.8, 1.2)
                        location.update_reputation(success, profit)
    
        def _detect_emergent_behaviors(self) -> Optional[Dict[str, Any]]:
            """Detect emergent behaviors in the ecosystem"""
    
            # Analyze current assignment patterns
            location_assignments = defaultdict(list)
            for product in self.product_agents:
                if product.current_location:
                    location_assignments[product.current_location].append(product)
    
            # Detect seasonal migration
            high_velocity_locations = [
                loc_id for loc_id, products in location_assignments.items()
                if sum(1 for p in products if p.is_high_velocity) / len(products) > 0.7
            ]
    
            if len(high_velocity_locations) > len(self.location_agents) * 0.1:  # More than 10% of locations
                return {
                    "type": "seasonal_migration",
                    "description": f"High-velocity products clustering in {len(high_velocity_locations)} locations",
                    "timestamp": self.current_time,
                    "affected_locations": len(high_velocity_locations)
                }
    
            # Detect affinity clustering
            category_clusters = defaultdict(list)
            for loc_id, products in location_assignments.items():
                if len(products) >= 5:  # Only consider locations with 5+ products
                    categories = [p.product_category for p in products]
                    dominant_category = max(set(categories), key=categories.count)
                    if categories.count(dominant_category) / len(categories) > 0.6:  # 60% same category
                        category_clusters[dominant_category].append(loc_id)
    
            if len(category_clusters) > 3:  # Multiple category clusters
                return {
                    "type": "affinity_clustering",
                    "description": f"Products self-organizing into {len(category_clusters)} category clusters",
                    "timestamp": self.current_time,
                    "clusters": dict(category_clusters)
                }
    
            return None
    
        def _calculate_performance(self) -> Dict[str, float]:
            """Calculate ecosystem performance metrics"""
    
            # Assignment rate
            assigned_products = sum(1 for p in self.product_agents if p.current_location is not None)
            assignment_rate = assigned_products / len(self.product_agents)
    
            # Load balancing
            location_loads = [l.current_load / l.capability.capacity for l in self.location_agents]
            load_balance = 1.0 - np.std(location_loads)  # Lower std = better balance
    
            # Market efficiency
            market_stats = self.market.get_market_stats()
            market_efficiency = market_stats.get("market_efficiency", 0.5)
    
            # Overall efficiency (weighted combination)
            efficiency = (
                0.4 * assignment_rate +
                0.3 * load_balance +
                0.3 * market_efficiency
            )
    
            return {
                "assignment_rate": assignment_rate,
                "load_balance": load_balance,
                "market_efficiency": market_efficiency,
                "efficiency": efficiency
            }
    
        def get_ecosystem_summary(self) -> Dict[str, Any]:
            """Get comprehensive ecosystem summary"""
    
            if not self.performance_history:
                return {"status": "no_simulation_data"}
    
            final_performance = self.performance_history[-1]["performance"]
            initial_performance = self.performance_history[0]["performance"]
    
            # Calculate improvements
            efficiency_improvement = final_performance["efficiency"] - initial_performance["efficiency"]
    
            # Agent statistics
            active_locations = sum(1 for l in self.location_agents if l.state == AgentState.ACTIVE)
            avg_reputation = np.mean([l.reputation_score for l in self.location_agents])
    
            # Contract statistics
            total_contracts = sum(len(p.contracts) for p in self.product_agents)
            avg_contract_value = 0
            if total_contracts > 0:
                all_contracts = [c for p in self.product_agents for c in p.contracts]
                avg_contract_value = np.mean([c.terms["price"] for c in all_contracts])
    
            # Emergent behavior summary
            behavior_counts = defaultdict(int)
            for behavior in self.emergent_behaviors:
                behavior_counts[behavior["type"]] += 1
    
            return {
                "simulation_cycles": len(self.performance_history),
                "total_contracts": total_contracts,
                "avg_contract_value": avg_contract_value,
                "final_efficiency": final_performance["efficiency"],
                "efficiency_improvement": efficiency_improvement,
                "assignment_rate": final_performance["assignment_rate"],
                "load_balance": final_performance["load_balance"],
                "market_efficiency": final_performance["market_efficiency"],
                "active_locations": active_locations,
                "avg_reputation": avg_reputation,
                "emergent_behaviors": dict(behavior_counts),
                "market_stats": self.market.get_market_stats()
            }
    
    # Create and run the autonomous ecosystem
    print("Initializing Autonomous Self-Optimizing Ecosystem...")
    ecosystem = AutonomousEcosystem(
        n_locations=500,      # 500 location agents
        n_products=100000     # 100,000 product agents
    )
    
    # Run Black Friday scenario
    ecosystem.run_simulation(duration_minutes=30, demand_scenario="black_friday")
except Exception as e:
    print(f'Error in cell: {e}')
```
