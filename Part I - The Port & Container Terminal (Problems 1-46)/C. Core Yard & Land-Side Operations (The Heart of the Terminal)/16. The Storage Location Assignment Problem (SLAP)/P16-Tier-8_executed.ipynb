5. **Algorithmic Auditing**: Continuous monitoring for bias and value drift
6. **Transparency Reporting**: Generate explainable decision documentation

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass
from typing import List, Tuple, Dict, Optional, Any
import random
import time
from datetime import datetime, timedelta
from collections import defaultdict
from enum import Enum
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
class StakeholderType(Enum):
    """Types of stakeholders in the warehouse ecosystem"""
    CUSTOMER = "customer"
    WORKER = "worker"
    MANAGEMENT = "management"
    ENVIRONMENT = "environment"
    SUPPLIER = "supplier"
    COMMUNITY = "community"

@dataclass
class Stakeholder:
    """Represents a stakeholder in the warehouse system"""
    stakeholder_id: str
    stakeholder_type: StakeholderType
    utility_function: Dict[str, float]  # weights for different objectives
    constraints: Dict[str, Any]
    priority: float  # stakeholder priority weight

@dataclass
class EthicalConstraint:
    """Represents an ethical constraint in the optimization"""
    constraint_id: str
    name: str
    description: str
    constraint_type: str  # 'hard', 'soft', 'preference'
    metric_function: str
    threshold: float
    penalty_weight: float
    audit_frequency: str  # 'continuous', 'daily', 'weekly'

@dataclass
class ValueFunction:
    """Multi-objective value function for optimization"""
    cost_efficiency_weight: float = 0.4
    worker_welfare_weight: float = 0.25
    environmental_impact_weight: float = 0.15
    service_equity_weight: float = 0.1
    supplier_fairness_weight: float = 0.05
    community_impact_weight: float = 0.05

@dataclass
class AssignmentDecision:
    """Represents a storage assignment decision with ethical justification"""
    product_id: str
    location_id: str
    cost_score: float
    ethical_scores: Dict[str, float]
    total_value_score: float
    justification: str
    stakeholder_impacts: Dict[str, float]
    timestamp: datetime
```

```python
class EthicalConstraintArchitecture:
    """Implements ethical constraints and constitutional AI principles"""
    
    def __init__(self):
        self.constraints = []
        self.audit_history = []
        self.violation_log = []
        self._initialize_constraints()
    
    def _initialize_constraints(self):
        """Initialize ethical constraints based on stakeholder requirements"""
        
        constraints = [
            # Worker welfare constraints
            EthicalConstraint(
                constraint_id="WORKER_WEIGHT_EQUITY",
                name="Worker Weight Equity",
                description="No worker handles more than 110% of average pick weight",
                constraint_type="hard",
                metric_function="max_weight_ratio",
                threshold=1.1,
                penalty_weight=1000.0,
                audit_frequency="continuous"
            ),
            
            # Service equity constraints
            EthicalConstraint(
                constraint_id="SERVICE_EQUITY",
                name="Service Level Equity",
                description="Maintain service level parity within 2% across customer segments",
                constraint_type="hard",
                metric_function="service_level_variance",
                threshold=0.02,
                penalty_weight=500.0,
                audit_frequency="daily"
            ),
            
            # Environmental constraints
            EthicalConstraint(
                constraint_id="ENERGY_REDUCTION",
                name="Energy Consumption Reduction",
                description="Reduce total energy consumption by 15% compared to baseline",
                constraint_type="soft",
                metric_function="energy_reduction_ratio",
                threshold=0.15,
                penalty_weight=200.0,
                audit_frequency="weekly"
            ),
            
            # Supplier fairness constraints
            EthicalConstraint(
                constraint_id="SUPPLIER_NEUTRALITY",
                name="Supplier Treatment Neutrality",
                description="Unbiased treatment of supplier products in assignment decisions",
                constraint_type="soft",
                metric_function="supplier_assignment_variance",
                threshold=0.1,
                penalty_weight=100.0,
                audit_frequency="weekly"
            ),
            
            # Community impact constraints
            EthicalConstraint(
                constraint_id="COMMUNITY_NOISE",
                name="Community Noise Impact",
                description="Limit noise pollution in nearby residential areas",
                constraint_type="preference",
                metric_function="noise_impact_score",
                threshold=0.3,
                penalty_weight=50.0,
                audit_frequency="continuous"
            ),
            
            # Workload distribution constraints
            EthicalConstraint(
                constraint_id="WORKLOAD_DISTRIBUTION",
                name="Fair Workload Distribution",
                description="Ensure fair distribution of picking effort across workers",
                constraint_type="hard",
                metric_function="workload_variance",
                threshold=0.15,
                penalty_weight=300.0,
                audit_frequency="daily"
            )
        ]
        
        self.constraints = constraints
    
    def evaluate_constraint_compliance(self, assignment_decisions: List[AssignmentDecision], 
                                      system_state: Dict[str, Any]) -> Dict[str, Any]:
        """Evaluate compliance with all ethical constraints"""
        
        compliance_results = {}
        total_penalty = 0.0
        violations = []
        
        for constraint in self.constraints:
            metric_value = self._calculate_metric(constraint.metric_function, assignment_decisions, system_state)
            
            is_compliant = metric_value <= constraint.threshold
            penalty = 0.0 if is_compliant else (metric_value - constraint.threshold) * constraint.penalty_weight
            
            compliance_results[constraint.constraint_id] = {
                "name": constraint.name,
                "metric_value": metric_value,
                "threshold": constraint.threshold,
                "compliant": is_compliant,
                "penalty": penalty,
                "severity": "high" if penalty > 500 else "medium" if penalty > 100 else "low"
            }
            
            total_penalty += penalty
            
            if not is_compliant:
                violations.append({
                    "constraint_id": constraint.constraint_id,
                    "name": constraint.name,
                    "metric_value": metric_value,
                    "threshold": constraint.threshold,
                    "penalty": penalty,
                    "timestamp": datetime.now()
                })
        
        # Record audit
        audit_record = {
            "timestamp": datetime.now(),
            "total_penalty": total_penalty,
            "compliance_rate": sum(1 for r in compliance_results.values() if r["compliant"]) / len(compliance_results),
            "violations_count": len(violations),
            "constraint_results": compliance_results
        }
        
        self.audit_history.append(audit_record)
        
        if violations:
            self.violation_log.extend(violations)
        
        return {
            "total_penalty": total_penalty,
            "compliance_rate": audit_record["compliance_rate"],
            "violations": violations,
            "constraint_results": compliance_results
        }
    
    def _calculate_metric(self, metric_function: str, assignment_decisions: List[AssignmentDecision], 
                         system_state: Dict[str, Any]) -> float:
        """Calculate specific metric for constraint evaluation"""
        
        if metric_function == "max_weight_ratio":
            # Calculate maximum worker weight ratio
            worker_weights = system_state.get("worker_weights", [])
            if not worker_weights:
                return 0.0
            avg_weight = np.mean(worker_weights)
            max_weight = max(worker_weights)
            return max_weight / avg_weight if avg_weight > 0 else 0.0
        
        elif metric_function == "service_level_variance":
            # Calculate service level variance across customer segments
            service_levels = system_state.get("service_levels", [])
            if len(service_levels) < 2:
                return 0.0
            return np.var(service_levels)
        
        elif metric_function == "energy_reduction_ratio":
            # Calculate energy reduction compared to baseline
            current_energy = system_state.get("current_energy_consumption", 1000)
            baseline_energy = system_state.get("baseline_energy_consumption", 1000)
            return (baseline_energy - current_energy) / baseline_energy if baseline_energy > 0 else 0.0
        
        elif metric_function == "supplier_assignment_variance":
            # Calculate variance in supplier treatment
            supplier_assignments = system_state.get("supplier_assignments", {})
            if len(supplier_assignments) < 2:
                return 0.0
            assignment_counts = list(supplier_assignments.values())
            return np.var(assignment_counts) / np.mean(assignment_counts) if np.mean(assignment_counts) > 0 else 0.0
        
        elif metric_function == "noise_impact_score":
            # Calculate noise impact on community
            high_noise_locations = system_state.get("high_noise_locations", 0)
            total_locations = system_state.get("total_locations", 100)
            return high_noise_locations / total_locations
        
        elif metric_function == "workload_variance":
            # Calculate workload variance across workers
            workloads = system_state.get("worker_workloads", [])
            if len(workloads) < 2:
                return 0.0
            return np.var(workloads) / np.mean(workloads) if np.mean(workloads) > 0 else 0.0
        
        else:
            return 0.0  # Unknown metric function
    
    def generate_transparency_report(self) -> Dict[str, Any]:
        """Generate transparency report for algorithmic governance"""
        
        if not self.audit_history:
            return {"status": "no_audit_data"}
        
        recent_audits = self.audit_history[-10:]  # Last 10 audits
        
        # Calculate statistics
        avg_compliance = np.mean([a["compliance_rate"] for a in recent_audits])
        avg_penalty = np.mean([a["total_penalty"] for a in recent_audits])
        total_violations = sum([a["violations_count"] for a in recent_audits])
        
        # Constraint-specific compliance
        constraint_compliance = {}
        for constraint in self.constraints:
            constraint_audits = []
            for audit in recent_audits:
                if constraint.constraint_id in audit["constraint_results"]:
                    constraint_audits.append(audit["constraint_results"][constraint.constraint_id]["compliant"])
            
            if constraint_audits:
                compliance_rate = sum(constraint_audits) / len(constraint_audits)
                constraint_compliance[constraint.constraint_id] = {
                    "name": constraint.name,
                    "compliance_rate": compliance_rate,
                    "type": constraint.constraint_type,
                    "status": "good" if compliance_rate >= 0.95 else "needs_attention" if compliance_rate >= 0.8 else "critical"
                }
        
        return {
            "report_timestamp": datetime.now(),
            "audit_period": f"Last {len(recent_audits)} audits",
            "overall_compliance": avg_compliance,
            "average_penalty": avg_penalty,
            "total_violations": total_violations,
            "constraint_compliance": constraint_compliance,
            "recommendations": self._generate_recommendations(constraint_compliance)
        }
    
    def _generate_recommendations(self, constraint_compliance: Dict[str, Any]) -> List[str]:
        """Generate recommendations based on compliance analysis"""
        
        recommendations = []
        
        for constraint_id, compliance_data in constraint_compliance.items():
            if compliance_data["status"] == "critical":
                recommendations.append(f"URGENT: Address {compliance_data['name']} compliance issues")
            elif compliance_data["status"] == "needs_attention":
                recommendations.append(f"Monitor {compliance_data['name']} compliance closely")
        
        if not recommendations:
            recommendations.append("All ethical constraints performing well")
        
        return recommendations
```

```python
class ValueAlignedOptimizer:
    """Multi-objective optimizer with ethical constraints"""
    
    def __init__(self, value_function: ValueFunction, ethical_architecture: EthicalConstraintArchitecture):
        self.value_function = value_function
        self.ethical_architecture = ethical_architecture
        
        # Optimization history
        self.optimization_history = []
        self.decision_log = []
    
    def optimize_assignment(self, products: List[Dict], locations: List[Dict], 
                           stakeholders: List[Stakeholder], system_state: Dict[str, Any]) -> List[AssignmentDecision]:
        """Perform value-aligned optimization with ethical constraints"""
        
        print("Performing Value-Aligned SLAP Optimization...")
        
        # Step 1: Calculate traditional cost-only assignment (baseline)
        baseline_assignments = self._calculate_baseline_assignments(products, locations)
        
        # Step 2: Evaluate baseline against ethical constraints
        baseline_compliance = self.ethical_architecture.evaluate_constraint_compliance(
            baseline_assignments, system_state
        )
        
        print(f"Baseline compliance: {baseline_compliance['compliance_rate']:.2%}")
        print(f"Baseline penalty: ${baseline_compliance['total_penalty']:.2f}")
        
        # Step 3: Perform ethical optimization
        ethical_assignments = self._perform_ethical_optimization(
            products, locations, stakeholders, system_state, baseline_assignments
        )
        
        # Step 4: Evaluate ethical assignment
        ethical_compliance = self.ethical_architecture.evaluate_constraint_compliance(
            ethical_assignments, system_state
        )
        
        print(f"Ethical compliance: {ethical_compliance['compliance_rate']:.2%}")
        print(f"Ethical penalty: ${ethical_compliance['total_penalty']:.2f}")
        
        # Step 5: Generate decision justifications
        for decision in ethical_assignments:
            decision.justification = self._generate_decision_justification(decision, stakeholders)
        
        # Record optimization
        optimization_record = {
            "timestamp": datetime.now(),
            "baseline_cost": sum(d.cost_score for d in baseline_assignments),
            "ethical_cost": sum(d.cost_score for d in ethical_assignments),
            "baseline_compliance": baseline_compliance["compliance_rate"],
            "ethical_compliance": ethical_compliance["compliance_rate"],
            "baseline_penalty": baseline_compliance["total_penalty"],
            "ethical_penalty": ethical_compliance["total_penalty"],
            "value_improvement": self._calculate_value_improvement(baseline_assignments, ethical_assignments)
        }
        
        self.optimization_history.append(optimization_record)
        self.decision_log.extend(ethical_assignments)
        
        return ethical_assignments
    
    def _calculate_baseline_assignments(self, products: List[Dict], locations: List[Dict]) -> List[AssignmentDecision]:
        """Calculate cost-optimal assignments (baseline)"""
        
        assignments = []
        
        for product in products:
            best_location = None
            best_cost = float('inf')
            
            for location in locations:
                # Simple cost calculation: distance * velocity
                cost = location["distance"] * product["velocity"]
                
                if cost < best_cost:
                    best_cost = cost
                    best_location = location["id"]
            
            if best_location:
                decision = AssignmentDecision(
                    product_id=product["id"],
                    location_id=best_location,
                    cost_score=best_cost,
                    ethical_scores={},
                    total_value_score=best_cost,  # Pure cost optimization
                    justification="Cost-optimal assignment",
                    stakeholder_impacts={},
                    timestamp=datetime.now()
                )
                assignments.append(decision)
        
        return assignments
    
    def _perform_ethical_optimization(self, products: List[Dict], locations: List[Dict], 
                                   stakeholders: List[Stakeholder], system_state: Dict[str, Any],
                                   baseline_assignments: List[AssignmentDecision]) -> List[AssignmentDecision]:
        """Perform optimization with ethical constraints"""
        
        assignments = []
        
        for product in products:
            best_decision = None
            best_value_score = -float('inf')
            
            for location in locations:
                # Calculate base cost
                cost = location["distance"] * product["velocity"]
                
                # Calculate ethical scores
                ethical_scores = self._calculate_ethical_scores(
                    product, location, stakeholders, system_state
                )
                
                # Calculate total value score (weighted combination)
                total_value = (
                    self.value_function.cost_efficiency_weight * cost +
                    self.value_function.worker_welfare_weight * ethical_scores["worker_welfare"] +
                    self.value_function.environmental_impact_weight * ethical_scores["environmental_impact"] +
                    self.value_function.service_equity_weight * ethical_scores["service_equity"] +
                    self.value_function.supplier_fairness_weight * ethical_scores["supplier_fairness"] +
                    self.value_function.community_impact_weight * ethical_scores["community_impact"]
                )
                
                # Apply ethical penalty if constraints would be violated
                test_decision = AssignmentDecision(
                    product_id=product["id"],
                    location_id=location["id"],
                    cost_score=cost,
                    ethical_scores=ethical_scores,
                    total_value_score=total_value,
                    justification="",
                    stakeholder_impacts={},
                    timestamp=datetime.now()
                )
                
                # Check if this decision would violate constraints
                test_assignments = assignments + [test_decision]
                compliance_result = self.ethical_architecture.evaluate_constraint_compliance(
                    test_assignments, system_state
                )
                
                # Apply penalty for constraint violations
                if compliance_result["total_penalty"] > 0:
                    total_value += compliance_result["total_penalty"]
                
                if total_value > best_value_score:
                    best_value_score = total_value
                    best_decision = test_decision
            
            if best_decision:
                assignments.append(best_decision)
        
        return assignments
    
    def _calculate_ethical_scores(self, product: Dict, location: Dict, 
                                 stakeholders: List[Stakeholder], system_state: Dict[str, Any]) -> Dict[str, float]:
        """Calculate ethical scores for a product-location assignment"""
        
        scores = {
            "worker_welfare": 0.0,
            "environmental_impact": 0.0,
            "service_equity": 0.0,
            "supplier_fairness": 0.0,
            "community_impact": 0.0
        }
        
        # Worker welfare score (based on location characteristics)
        if location.get("ergonomic_score", 0) > 0:
            scores["worker_welfare"] = location["ergonomic_score"]
        else:
            # Simulate ergonomic score based on distance and accessibility
            distance_factor = 1.0 - (location["distance"] / 100.0)  # Closer = better ergonomics
            scores["worker_welfare"] = max(0, min(1, distance_factor))
        
        # Environmental impact score (based on energy consumption)
        if location.get("energy_efficiency", 0) > 0:
            scores["environmental_impact"] = location["energy_efficiency"]
        else:
            # Simulate energy efficiency (closer = less energy for transport)
            energy_score = 1.0 - (location["distance"] / 50.0)  # Closer = more energy efficient
            scores["environmental_impact"] = max(0, min(1, energy_score))
        
        # Service equity score (based on service level consistency)
        customer_segment = product.get("customer_segment", "standard")
        if customer_segment == "premium":
            scores["service_equity"] = 0.8  # Slightly lower to promote equity
        elif customer_segment == "economy":
            scores["service_equity"] = 1.0  # Higher to promote equity
        else:
            scores["service_equity"] = 0.9  # Standard
        
        # Supplier fairness score (based on supplier neutrality)
        supplier_rank = product.get("supplier_rank", 5)  # 1-10 ranking
        scores["supplier_fairness"] = 1.0 - abs(supplier_rank - 5.5) / 4.5  # Middle ranks get higher scores
        
        # Community impact score (based on noise and local impact)
        if location.get("near_residential", False):
            scores["community_impact"] = 0.3  # Lower impact for residential areas
        else:
            scores["community_impact"] = 0.8  # Higher impact for industrial areas
        
        return scores
    
    def _generate_decision_justification(self, decision: AssignmentDecision, stakeholders: List[Stakeholder]) -> str:
        """Generate explanation for assignment decision"""
        
        justification_parts = []
        
        # Cost justification
        justification_parts.append(f"Cost efficiency: ${decision.cost_score:.2f}")
        
        # Ethical justifications
        if decision.ethical_scores.get("worker_welfare", 0) > 0.7:
            justification_parts.append("High worker welfare score")
        
        if decision.ethical_scores.get("environmental_impact", 0) > 0.7:
            justification_parts.append("Low environmental impact")
        
        if decision.ethical_scores.get("service_equity", 0) > 0.8:
            justification_parts.append("Promotes service equity")
        
        if decision.ethical_scores.get("supplier_fairness", 0) > 0.7:
            justification_parts.append("Supplier neutral assignment")
        
        if decision.ethical_scores.get("community_impact", 0) > 0.7:
            justification_parts.append("Minimal community impact")
        
        return "; ".join(justification_parts)
    
    def _calculate_value_improvement(self, baseline_assignments: List[AssignmentDecision], 
                                   ethical_assignments: List[AssignmentDecision]) -> Dict[str, float]:
        """Calculate value improvement metrics"""
        
        baseline_cost = sum(d.cost_score for d in baseline_assignments)
        ethical_cost = sum(d.cost_score for d in ethical_assignments)
        
        cost_increase = (ethical_cost - baseline_cost) / baseline_cost * 100
        
        # Calculate ethical score improvements
        baseline_ethical_avg = 0.5  # Assume baseline has average ethical performance
        ethical_ethical_avg = np.mean([
            np.mean(list(d.ethical_scores.values())) if d.ethical_scores else 0.5
            for d in ethical_assignments
        ])
        
        ethical_improvement = (ethical_ethical_avg - baseline_ethical_avg) / baseline_ethical_avg * 100
        
        return {
            "cost_increase_percentage": cost_increase,
            "ethical_improvement_percentage": ethical_improvement,
            "value_score_improvement": ethical_improvement - cost_increase
        }
    
    def generate_optimization_report(self) -> Dict[str, Any]:
        """Generate comprehensive optimization report"""
        
        if not self.optimization_history:
            return {"status": "no_optimization_data"}
        
        latest = self.optimization_history[-1]
        
        # Calculate trends
        if len(self.optimization_history) >= 5:
            recent_costs = [opt["ethical_cost"] for opt in self.optimization_history[-5:]]
            cost_trend = "decreasing" if recent_costs[-1] < recent_costs[0] else "increasing"
            
            recent_compliance = [opt["ethical_compliance"] for opt in self.optimization_history[-5:]]
            compliance_trend = "improving" if recent_compliance[-1] > recent_compliance[0] else "declining"
        else:
            cost_trend = "insufficient_data"
            compliance_trend = "insufficient_data"
        
        return {
            "latest_optimization": latest,
            "total_optimizations": len(self.optimization_history),
            "total_decisions": len(self.decision_log),
            "cost_trend": cost_trend,
            "compliance_trend": compliance_trend,
            "transparency_report": self.ethical_architecture.generate_transparency_report()
        }
```

```python
# Visualization of Value-Aligned Optimization results
def visualize_value_aligned_results(optimizer: ValueAlignedOptimizer, ethical_architecture: EthicalConstraintArchitecture):
    """Create comprehensive visualization of value-aligned results"""
    
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    fig.suptitle('Value-Aligned SLAP Optimization - Ethical Framework Results', fontsize=16, fontweight='bold')
    
    # 1. Cost vs Ethical Performance Comparison
    ax1 = axes[0, 0]
    if len(optimizer.optimization_history) >= 1:
        latest = optimizer.optimization_history[-1]
        
        methods = ['Pure Cost Opt.', 'Value-Aligned']
        costs = [latest["baseline_cost"], latest["ethical_cost"]]
        colors = ['red', 'green']
        
        bars = ax1.bar(methods, costs, color=colors, alpha=0.7)
        ax1.set_title('Cost Comparison')
        ax1.set_ylabel('Total Cost ($)')
        ax1.grid(True, alpha=0.3)
        
        # Add value labels
        for bar, cost in zip(bars, costs):
            height = bar.get_height()
            ax1.text(bar.get_x() + bar.get_width()/2., height + max(costs)*0.01,
                    f'${cost:,.0f}', ha='center', va='bottom')
        
        # Add savings annotation
        savings = latest["baseline_cost"] - latest["ethical_cost"]
        savings_pct = (savings / latest["baseline_cost"]) * 100
        ax1.annotate(f'Savings: ${savings:,.0f} ({savings_pct:.1f}%)', 
                    xy=(1, latest["ethical_cost"]),
                    xytext=(10, 10), textcoords='offset points',
                    bbox=dict(boxstyle='round,pad=0.5', fc='lightgreen', alpha=0.7),
                    arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0'))
    
    # 2. Compliance Rate Comparison
    ax2 = axes[0, 1]
    if len(optimizer.optimization_history) >= 1:
        latest = optimizer.optimization_history[-1]
        
        methods = ['Baseline', 'Value-Aligned']
        compliance_rates = [latest["baseline_compliance"], latest["ethical_compliance"]]
        colors = ['orange', 'blue']
        
        bars = ax2.bar(methods, compliance_rates, color=colors, alpha=0.7)
        ax2.set_title('Ethical Compliance Rate')
        ax2.set_ylabel('Compliance Rate')
        ax2.set_ylim(0, 1)
        ax2.grid(True, alpha=0.3)
        
        # Add percentage labels
        for bar, rate in zip(bars, compliance_rates):
            height = bar.get_height()
            ax2.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                    f'{rate:.1%}', ha='center', va='bottom')
    
    # 3. Ethical Constraint Compliance
    ax3 = axes[0, 2]
    transparency_report = ethical_architecture.generate_transparency_report()
    
    if "constraint_compliance" in transparency_report:
        constraint_data = transparency_report["constraint_compliance"]
        constraint_names = [data["name"][:20] for data in constraint_data.values()]  # Truncate long names
        compliance_rates = [data["compliance_rate"] for data in constraint_data.values()]
        
        colors = ['green' if rate >= 0.95 else 'yellow' if rate >= 0.8 else 'red' for rate in compliance_rates]
        bars = ax3.barh(constraint_names, compliance_rates, color=colors, alpha=0.7)
        ax3.set_title('Constraint Compliance Rates')
        ax3.set_xlabel('Compliance Rate')
        ax3.set_xlim(0, 1)
        ax3.grid(True, alpha=0.3)
    
    # 4. Stakeholder Priority Distribution
    ax4 = axes[1, 0]
    stakeholders = [s for s in optimizer.decision_log[0].stakeholder_impacts.keys()] if optimizer.decision_log else []
    if stakeholders:
        # Use the stakeholder data from our example
        stakeholder_priorities = {
            'Customer': 0.30,
            'Worker': 0.25,
            'Management': 0.20,
            'Environment': 0.15,
            'Supplier': 0.05,
            'Community': 0.05
        }
        
        names = list(stakeholder_priorities.keys())
        priorities = list(stakeholder_priorities.values())
        
        colors = plt.cm.Set3(np.linspace(0, 1, len(names)))
        wedges, texts, autotexts = ax4.pie(priorities, labels=names, colors=colors, autopct='%1.1f%%', startangle=90)
        ax4.set_title('Stakeholder Priority Distribution')
    
    # 5. Ethical Score Distribution
    ax5 = axes[1, 1]
    if optimizer.decision_log:
        # Collect ethical scores
        all_ethical_scores = []
        for decision in optimizer.decision_log:
            all_ethical_scores.extend(decision.ethical_scores.values())
        
        ax5.hist(all_ethical_scores, bins=20, alpha=0.7, color='skyblue', edgecolor='black')
        ax5.set_title('Ethical Score Distribution')
        ax5.set_xlabel('Ethical Score')
        ax5.set_ylabel('Frequency')
        ax5.grid(True, alpha=0.3)
        
        # Add mean line
        mean_score = np.mean(all_ethical_scores)
        ax5.axvline(mean_score, color='red', linestyle='--', 
                   label=f'Mean: {mean_score:.3f}')
        ax5.legend()
    
    # 6. Value Function Weight Analysis
    ax6 = axes[1, 2]
    value_weights = {
        'Cost Efficiency': optimizer.value_function.cost_efficiency_weight,
        'Worker Welfare': optimizer.value_function.worker_welfare_weight,
        'Environmental': optimizer.value_function.environmental_impact_weight,
        'Service Equity': optimizer.value_function.service_equity_weight,
        'Supplier Fairness': optimizer.value_function.supplier_fairness_weight,
        'Community Impact': optimizer.value_function.community_impact_weight
    }
    
    names = list(value_weights.keys())
    weights = list(value_weights.values())
    
    colors = plt.cm.viridis(np.linspace(0, 1, len(names)))
    bars = ax6.bar(names, weights, color=colors, alpha=0.7)
    ax6.set_title('Value Function Weights')
    ax6.set_ylabel('Weight')
    ax6.tick_params(axis='x', rotation=45)
    ax6.grid(True, alpha=0.3)
    
    # Add value labels
    for bar, weight in zip(bars, weights):
        height = bar.get_height()
        ax6.text(bar.get_x() + bar.get_width()/2., height + max(weights)*0.01,
                f'{weight:.2f}', ha='center', va='bottom')
    
    plt.tight_layout()
    plt.show()
    
    # Print comprehensive summary
    print(f"\n{'='*80}")
    print(f"VALUE-ALIGNED SLAP OPTIMIZATION SUMMARY")
    print(f"{'='*80}")
    
    if optimizer.optimization_history:
        latest = optimizer.optimization_history[-1]
        value_improvement = latest["value_improvement"]
        
        print(f"\nFinancial Performance:")
        print(f"  Pure cost optimization: ${latest['baseline_cost']:,.2f}")
        print(f"  Value-aligned optimization: ${latest['ethical_cost']:,.2f}")
        print(f"  Cost savings: ${latest['baseline_cost'] - latest['ethical_cost']:,.2f}")
        print(f"  Cost reduction: {((latest['baseline_cost'] - latest['ethical_cost']) / latest['baseline_cost']) * 100:.1f}%")
        
        print(f"\nEthical Performance:")
        print(f"  Baseline compliance: {latest['baseline_compliance']:.1%}")
        print(f"  Value-aligned compliance: {latest['ethical_compliance']:.1%}")
        print(f"  Compliance improvement: {(latest['ethical_compliance'] - latest['baseline_compliance']) * 100:.1f}%")
        print(f"  Ethical penalty reduction: ${latest['baseline_penalty'] - latest['ethical_penalty']:,.2f}")
        
        print(f"\nValue Improvement:")
        print(f"  Cost increase: {value_improvement['cost_increase_percentage']:.1f}%")
        print(f"  Ethical improvement: {value_improvement['ethical_improvement_percentage']:.1f}%")
        print(f"  Net value gain: {value_improvement['value_score_improvement']:.1f}%")
        
        # Validation against source expectations
        print(f"\nValidation Against Source Expectations:")
        expected_savings = 300000  # $2.1M - $1.8M
        actual_savings = latest['baseline_cost'] - latest['ethical_cost']
        
        savings_achieved = actual_savings >= expected_savings * 0.8  # 80% of expected
        compliance_improved = latest['ethical_compliance'] >= 0.95
        
        print(f"  Expected savings: ${expected_savings:,.0f}")
        print(f"  Actual savings: ${actual_savings:,.0f}")
        print(f"  Savings target: {'✓ ACHIEVED' if savings_achieved else '✗ NOT ACHIEVED'}")
        print(f"  High compliance: {'✓ ACHIEVED' if compliance_improved else '✗ NEEDS IMPROVEMENT'}")
        print(f"  Overall assessment: {'✓ EXCELLENT' if savings_achieved and compliance_improved else '✓ GOOD' if savings_achieved or compliance_improved else '✗ NEEDS IMPROVEMENT'}")
    
    # Transparency report summary
    transparency_report = ethical_architecture.generate_transparency_report()
    if "overall_compliance" in transparency_report:
        print(f"\nTransparency Report Summary:")
        print(f"  Overall compliance: {transparency_report['overall_compliance']:.1%}")
        print(f"  Average penalty: ${transparency_report['average_penalty']:.2f}")
        print(f"  Total violations: {transparency_report['total_violations']}")
        
        recommendations = transparency_report.get('recommendations', [])
        print(f"  Recommendations: {len(recommendations)}")
        for rec in recommendations[:3]:  # Show first 3
            print(f"    - {rec}")

# Visualize the results
visualize_value_aligned_results(optimizer, ethical_architecture)
```

```python
# Detailed ethical constraint analysis
def analyze_ethical_constraints(ethical_architecture: EthicalConstraintArchitecture, optimizer: ValueAlignedOptimizer):
    """Detailed analysis of ethical constraints and their impact"""
    
    print("\n" + "="*70)
    print("ETHICAL CONSTRAINT ANALYSIS")
    print("="*70)
    
    # Analyze each constraint
    print(f"\nConstraint Performance Analysis:")
    print(f"{'Constraint':<25} {'Type':<10} {'Threshold':<12} {'Current':<12} {'Status':<10} {'Penalty':<10}")
    print("-" * 80)
    
    if optimizer.optimization_history:
        latest_compliance = optimizer.optimization_history[-1].get("ethical_compliance_results", {})
        
        for constraint in ethical_architecture.constraints:
            constraint_id = constraint.constraint_id
            if constraint_id in latest_compliance:
                result = latest_compliance[constraint_id]
                
                status_icon = "✓" if result["compliant"] else "✗"
                
                print(f"{constraint.name:<25} {constraint.constraint_type:<10} {constraint.threshold:<12.2f} "
                      f"{result['metric_value']:<12.3f} {status_icon:<10} ${result['penalty']:<10.0f}")
    
    # Analyze constraint violations
    print(f"\nConstraint Violations Analysis:")
    
    violations_by_type = defaultdict(list)
    for violation in ethical_architecture.violation_log[-10:]:  # Last 10 violations
        constraint = next(c for c in ethical_architecture.constraints if c.constraint_id == violation["constraint_id"])
        violations_by_type[constraint.constraint_type].append(violation)
    
    for constraint_type, violations in violations_by_type.items():
        print(f"\n{constraint_type.title()} Constraints ({len(violations)} violations):")
        for violation in violations[:3]:  # Show first 3
            print(f"  {violation['name']}: {violation['metric_value']:.3f} > {violation['threshold']:.3f}")
            print(f"    Penalty: ${violation['penalty']:.2f}")
    
    # Analyze stakeholder impact
    print(f"\nStakeholder Impact Assessment:")
    
    # Calculate stakeholder satisfaction based on constraint compliance
    stakeholder_satisfaction = {
        "Customer": 0.0,
        "Worker": 0.0,
        "Management": 0.0,
        "Environment": 0.0,
        "Supplier": 0.0,
        "Community": 0.0
    }
    
    # Map constraints to stakeholders
    constraint_stakeholder_map = {
        "SERVICE_EQUITY": "Customer",
        "WORKER_WEIGHT_EQUITY": "Worker",
        "WORKLOAD_DISTRIBUTION": "Worker",
        "ENERGY_REDUCTION": "Environment",
        "SUPPLIER_NEUTRALITY": "Supplier",
        "COMMUNITY_NOISE": "Community"
    }
    
    if optimizer.optimization_history:
        latest_compliance = optimizer.optimization_history[-1].get("ethical_compliance_results", {})
        
        for constraint_id, constraint_data in latest_compliance.items():
            if constraint_id in constraint_stakeholder_map:
                stakeholder = constraint_stakeholder_map[constraint_id]
                compliance_score = 1.0 if constraint_data["compliant"] else 0.5
                stakeholder_satisfaction[stakeholder] += compliance_score
    
    # Normalize satisfaction scores
    for stakeholder in stakeholder_satisfaction:
        # Count relevant constraints per stakeholder
        relevant_constraints = sum(1 for c_id in constraint_stakeholder_map.values() if c_id == stakeholder)
        if relevant_constraints > 0:
            stakeholder_satisfaction[stakeholder] /= relevant_constraints
    
    print(f"{'Stakeholder':<15} {'Satisfaction':<12} {'Status':<15}")
    print("-" * 42)
    
    for stakeholder, satisfaction in stakeholder_satisfaction.items():
        status = "Excellent" if satisfaction >= 0.9 else "Good" if satisfaction >= 0.7 else "Needs Attention"
        print(f"{stakeholder:<15} {satisfaction:<12.1%} {status:<15}")
    
    # Business value vs ethical trade-off analysis
    print(f"\nBusiness vs Ethical Trade-off Analysis:")
    
    if len(optimizer.optimization_history) >= 1:
        latest = optimizer.optimization_history[-1]
        
        cost_efficiency_loss = ((latest["ethical_cost"] - latest["baseline_cost"]) / latest["baseline_cost"]) * 100
        ethical_compliance_gain = (latest["ethical_compliance"] - latest["baseline_compliance"]) * 100
        
        print(f"  Cost efficiency trade-off: {cost_efficiency_loss:.1f}% (negative = savings)")
        print(f"  Ethical compliance gain: {ethical_compliance_gain:.1f}%")
        print(f"  Net value proposition: {ethical_compliance_gain - cost_efficiency_loss:.1f}%")
        
        if ethical_compliance_gain > abs(cost_efficiency_loss):
            print(f"  Conclusion: Ethical optimization provides net positive value")
        else:
            print(f"  Conclusion: Ethical optimization requires investment in stakeholder value")
    
    # Algorithmic governance assessment
    print(f"\nAlgorithmic Governance Assessment:")
    
    transparency_report = ethical_architecture.generate_transparency_report()
    if "overall_compliance" in transparency_report:
        overall_compliance = transparency_report["overall_compliance"]
        
        print(f"  Algorithmic transparency: {'High' if overall_compliance >= 0.95 else 'Medium' if overall_compliance >= 0.8 else 'Low'}")
        print(f"  Bias detection: {'Active' if len(transparency_report['violations']) == 0 else f'{len(transparency_report["violations"])} issues detected'}")
        print(f"  Accountability: {'Strong' if transparency_report['total_violations'] == 0 else 'Needs Improvement'}")
        print(f"  Stakeholder trust: {'High' if overall_compliance >= 0.9 else 'Medium' if overall_compliance >= 0.7 else 'Low'}")
        
        # Governance recommendations
        print(f"\nGovernance Recommendations:")
        recommendations = transparency_report.get('recommendations', [])
        for i, rec in enumerate(recommendations[:5], 1):
            print(f"  {i}. {rec}")

# Analyze ethical constraints
analyze_ethical_constraints(ethical_architecture, optimizer)
```
