# 21. The Yard Block Housekeeping Problem

## Tier 7: Human-AI Symbiotic Partnership (The Centaur Model)

### Goal
Create a collaborative intelligence model where human expertise and AI capabilities combine synergistically for optimal yard block housekeeping decisions, leveraging the unique strengths of both human intuition and artificial intelligence.

### Key assumptions
- Human operators provide contextual awareness, creative problem-solving, and nuanced judgment
- AI systems offer computational power, pattern recognition, and optimization across vast solution spaces
- Transparent AI explanations enable effective human-AI collaboration
- Human insights can be captured to improve future AI recommendations
- Collaborative decisions outperform either humans or AI working alone

### Approach (step-by-step)
1. Design the symbiotic architecture with human and AI component roles
2. Implement Explainable AI (XAI) for transparent decision reasoning
3. Create collaborative decision-making interface with human oversight
4. Develop learning mechanisms to capture human insights and improve AI
5. Demonstrate collaborative housekeeping sessions with real-time interaction
6. Measure performance improvement over standalone AI or human approaches
7. Show emergent benefits from human-AI partnership

### What to look for in the results
- Transparent AI reasoning with clear explanation components
- Human-AI collaborative decision-making process with mutual learning
- Performance improvements over standalone approaches
- Captured human insights that enhance AI capabilities
- Real-time adaptation based on human feedback
- Emergent intelligent behavior from the partnership

### Concrete example (from the source)
A Tuesday morning scenario with weather uncertainty and premium customer constraints:
- **AI Initial Analysis**: 142 moves in 3.5-hour window, 97% completion, $8,400 cost, 82% confidence
- **Human Modification**: Reduced to 127 moves in 3-hour window, excluded premium customer containers
- **Collaborative Outcome**: 132 moves executed (104% of revised plan), weather buffer maintained
- **Learning Capture**: Weather forecast buffers increased 30%, premium customer sensitivity learned
- **Performance**: Optimal balance of efficiency, risk, and service with 100% customer satisfaction

### Why this Tier exists vs earlier Tiers
Tier 7 represents the evolution from automated systems to collaborative intelligence:
- **Human Wisdom Integration**: Incorporates experience, intuition, and contextual understanding
- **Explainable AI**: Provides transparent reasoning for human evaluation and augmentation
- **Mutual Learning**: Both human and AI learn from each other's strengths
- **Practical Implementation**: Addresses real-world constraints and exceptions
- **Adaptive Expertise**: Combines computational power with human judgment for optimal outcomes

### Pros / Cons vs Tiers 1-5
**Pros:**
- Superior decision quality through human-AI collaboration
- Handles exceptional situations and edge cases effectively
- Transparent and explainable decision-making process
- Continuous learning and improvement from human insights
- Practical applicability to real-world operations
- Higher user acceptance and trust in AI recommendations

**Cons:**
- Requires skilled human operators for effective collaboration
- Slower decision-making than fully automated systems
- Dependence on human availability and attention
- More complex implementation and user interface requirements
- Potential for human bias in decision-making

### When to use this Tier
- Critical operations where human judgment is essential
- Environments with high uncertainty and exceptional situations
- When customer relationships and service quality are paramount
- Operations requiring ethical considerations and risk assessment
- When building trust and acceptance of AI systems in operations

```python
# Enhanced data structures for Human-AI Collaboration

class DecisionConfidence(Enum):
    """AI confidence levels for decisions"""
    VERY_LOW = "very_low"
    LOW = "low"
    MEDIUM = "medium"
    HIGH = "high"
    VERY_HIGH = "very_high"

class HumanFeedbackType(Enum):
    """Types of human feedback on AI recommendations"""
    ACCEPT = "accept"
    MODIFY = "modify"
    REJECT = "reject"
    REQUEST_ALTERNATIVES = "request_alternatives"
    PROVIDE_CONTEXT = "provide_context"

@dataclass
class AIExplanation:
    """Explainable AI reasoning components"""
    primary_reasoning: str
    secondary_reasoning: str
    confidence_level: float  # 0.0 to 1.0
    alternatives: List[Dict[str, Any]]
    risks: List[str]
    data_sources: List[str]
    uncertainty_factors: List[str]
    expected_outcomes: Dict[str, float]

@dataclass
class HumanInsight:
    """Human operator insights and expertise"""
    insight_type: str
    context: str
    reasoning: str
    confidence: float
    priority: float
    applicable_conditions: List[str]
    learned_patterns: List[str]

@dataclass
class CollaborativeDecision:
    """Decision made through human-AI collaboration"""
    timestamp: datetime
    ai_recommendation: Dict[str, Any]
    ai_explanation: AIExplanation
    human_feedback: Optional[Dict[str, Any]]
    human_insights: List[HumanInsight]
    final_decision: Dict[str, Any]
    collaboration_quality: float  # 0.0 to 1.0
    learning_outcomes: List[str]

@dataclass
class CustomerConstraint:
    """Customer-specific constraints and preferences"""
    customer_id: str
    customer_name: str
    priority_level: str  # 'premium', 'standard', 'economy'
    handling_requirements: List[str]
    service_level_agreement: Dict[str, Any]
    special_instructions: List[str]
    cost_sensitivity: float
    time_sensitivity: float

@dataclass
class OperationalContext:
    """Current operational context for decision making"""
    timestamp: datetime
    weather_conditions: Dict[str, Any]
    equipment_status: Dict[str, str]
    staff_availability: Dict[str, int]
    vessel_schedule: List[Dict[str, Any]]
    gate_traffic_level: str
    maintenance_activities: List[str]
    safety_constraints: List[str]
    regulatory_requirements: List[str]

@dataclass
class LearningCapture:
    """Captured learning from human-AI interactions"""
    interaction_id: str
    timestamp: datetime
    situation_description: str
    human_modification: Dict[str, Any]
    original_ai_recommendation: Dict[str, Any]
    outcome_better: bool
    learning_extracted: str
    generalizable_pattern: bool
    confidence_improvement: float
```

```python
class ExplainableAI:
    """Explainable AI system for transparent housekeeping recommendations"""
    
    def __init__(self):
        self.decision_history = []
        self.pattern_library = {}
        self.explanation_templates = {
            'high_priority_vessel': {
                'primary': 'Preparing for high-priority vessel arrival',
                'secondary': 'Optimizing adjacent blocks to minimize vessel delays',
                'risks': ['Weather delay', 'Equipment failure', 'Gate congestion']
            },
            'weather_constraint': {
                'primary': 'Adjusting for adverse weather conditions',
                'secondary': 'Ensuring safety and operational continuity',
                'risks': ['Reduced equipment capacity', 'Extended operation times', 'Safety concerns']
            },
            'equipment_optimization': {
                'primary': 'Balancing equipment utilization across operations',
                'secondary': 'Minimizing idle time and maximizing throughput',
                'risks': ['Equipment bottlenecks', 'Maintenance conflicts', 'Operator fatigue']
            },
            'customer_service': {
                'primary': 'Maintaining service level agreements',
                'secondary': 'Protecting premium customer relationships',
                'risks': ['Customer dissatisfaction', 'SLA violations', 'Revenue loss']
            }
        }
    
    def generate_recommendation(self, yard_state: Dict, context: OperationalContext, 
                                 customer_constraints: List[CustomerConstraint]) -> Tuple[Dict, AIExplanation]:
        """Generate AI recommendation with detailed explanation"""
        
        # Analyze current situation
        situation_analysis = self._analyze_situation(yard_state, context, customer_constraints)
        
        # Generate base recommendation
        recommendation = self._generate_base_recommendation(situation_analysis)
        
        # Create detailed explanation
        explanation = self._create_explanation(recommendation, situation_analysis, context)
        
        return recommendation, explanation
    
    def _analyze_situation(self, yard_state: Dict, context: OperationalContext, 
                          customer_constraints: List[CustomerConstraint]) -> Dict:
        """Analyze current operational situation"""
        
        analysis = {
            'urgent_blocks': [],
            'available_blocks': [],
            'equipment_constraints': [],
            'customer_priorities': [],
            'weather_impact': 1.0,
            'time_pressure': 'normal',
            'risk_factors': [],
            'opportunities': []
        }
        
        # Analyze block utilization
        for block_id, block in yard_state.get('blocks', {}).items():
            if block.utilization > 0.9:
                analysis['urgent_blocks'].append(block_id)
            elif block.utilization < 0.7:
                analysis['available_blocks'].append(block_id)
        
        # Analyze equipment status
        for eq_id, status in context.equipment_status.items():
            if status != 'operational':
                analysis['equipment_constraints'].append(eq_id)
        
        # Analyze customer priorities
        for constraint in customer_constraints:
            if constraint.priority_level == 'premium':
                analysis['customer_priorities'].append(constraint.customer_id)
        
        # Analyze weather impact
        weather = context.weather_conditions
        if weather.get('wind_speed', 0) > 30:  # High wind
            analysis['weather_impact'] = 0.6
            analysis['risk_factors'].append('high_wind')
        elif weather.get('precipitation', 0) > 5:  # Heavy rain
            analysis['weather_impact'] = 0.8
            analysis['risk_factors'].append('heavy_rain')
        
        # Analyze time pressure
        upcoming_vessels = [v for v in context.vessel_schedule 
                          if v.get('eta', datetime.now()) <= context.timestamp + timedelta(hours=4)]
        
        if len(upcoming_vessels) >= 2:
            analysis['time_pressure'] = 'high'
            analysis['risk_factors'].append('vessel_congestion')
        elif len(upcoming_vessels) >= 1:
            analysis['time_pressure'] = 'medium'
        
        # Identify opportunities
        if context.gate_traffic_level == 'low':
            analysis['opportunities'].append('low_gate_traffic')
        
        if len(analysis['equipment_constraints']) == 0:
            analysis['opportunities'].append('full_equipment_availability')
        
        return analysis
    
    def _generate_base_recommendation(self, analysis: Dict) -> Dict:
        """Generate base housekeeping recommendation"""
        
        recommendation = {
            'moves': [],
            'time_window': '3_hours',
            'equipment_allocation': {},
            'priority_score': 0.0,
            'estimated_cost': 0.0,
            'expected_completion': 0.0,
            'risk_level': 'medium'
        }
        
        # Generate moves based on analysis
        move_count = 0
        
        # Priority 1: Handle urgent blocks
        for urgent_block in analysis['urgent_blocks'][:3]:  # Top 3 urgent blocks
            for available_block in analysis['available_blocks'][:2]:  # Top 2 available blocks
                if move_count < 10:  # Limit total moves
                    move_quantity = min(
                        20,  # Maximum 20 TEU per move
                        int(yard_state['blocks'][urgent_block].current_load * 0.1),
                        yard_state['blocks'][available_block].available_capacity // 2
                    )
                    
                    if move_quantity > 0:
                        move = {
                            'source_block': urgent_block,
                            'destination_block': available_block,
                            'quantity': move_quantity,
                            'priority': 'high',
                            'estimated_duration': move_quantity * 2,  # minutes
                            'cost': move_quantity * 15
                        }
                        recommendation['moves'].append(move)
                        recommendation['estimated_cost'] += move['cost']
                        move_count += 1
        
        # Calculate priority score
        urgency_factor = len(analysis['urgent_blocks']) * 0.3
        time_pressure_factor = {'high': 0.4, 'medium': 0.2, 'normal': 0.0}[analysis['time_pressure']]
        weather_factor = (1.0 - analysis['weather_impact']) * 0.2
        customer_factor = len(analysis['customer_priorities']) * 0.1
        
        recommendation['priority_score'] = urgency_factor + time_pressure_factor + weather_factor + customer_factor
        
        # Determine risk level
        if len(analysis['risk_factors']) >= 3 or analysis['weather_impact'] < 0.7:
            recommendation['risk_level'] = 'high'
        elif len(analysis['risk_factors']) >= 1:
            recommendation['risk_level'] = 'medium'
        else:
            recommendation['risk_level'] = 'low'
        
        # Estimate completion percentage
        total_urgent_capacity = sum(
            yard_state['blocks'][block_id].current_load - int(yard_state['blocks'][block_id].capacity * 0.8)
            for block_id in analysis['urgent_blocks']
        )
        
        total_moved = sum(move['quantity'] for move in recommendation['moves'])
        recommendation['expected_completion'] = min(1.0, total_moved / max(1, total_urgent_capacity))
        
        return recommendation
    
    def _create_explanation(self, recommendation: Dict, analysis: Dict, 
                          context: OperationalContext) -> AIExplanation:
        """Create detailed explanation for the recommendation"""
        
        # Determine primary reasoning
        if analysis['time_pressure'] == 'high':
            template_key = 'high_priority_vessel'
        elif analysis['weather_impact'] < 0.8:
            template_key = 'weather_constraint'
        elif len(analysis['customer_priorities']) > 0:
            template_key = 'customer_service'
        else:
            template_key = 'equipment_optimization'
        
        template = self.explanation_templates[template_key]
        
        # Build explanation
        primary_reasoning = template['primary']
        secondary_reasoning = template['secondary']
        
        # Add specific context
        if analysis['urgent_blocks']:
            primary_reasoning += f" (blocks {', '.join(map(str, analysis['urgent_blocks'][:3]))} over-utilized)"
        
        if analysis['time_pressure'] != 'normal':
            secondary_reasoning += f" ({analysis['time_pressure']} time pressure)"
        
        # Calculate confidence
        base_confidence = 0.8
        if len(analysis['risk_factors']) > 0:
            base_confidence -= 0.1 * len(analysis['risk_factors'])
        if analysis['weather_impact'] < 0.8:
            base_confidence -= 0.2
        if len(analysis['opportunities']) > 0:
            base_confidence += 0.1 * len(analysis['opportunities'])
        
        confidence = max(0.3, min(0.95, base_confidence))
        
        # Generate alternatives
        alternatives = []
        
        # Alternative 1: Reduced scope
        alt1_moves = recommendation['moves'][:len(recommendation['moves'])//2]
        alternatives.append({
            'description': 'Reduced scope - focus on highest priority moves only',
            'moves': len(alt1_moves),
            'completion': recommendation['expected_completion'] * 0.6,
            'cost': sum(move['cost'] for move in alt1_moves),
            'duration': sum(move['estimated_duration'] for move in alt1_moves)
        })
        
        # Alternative 2: Extended timeline
        alternatives.append({
            'description': 'Extended timeline - spread operations over 4 hours',
            'moves': len(recommendation['moves']),
            'completion': recommendation['expected_completion'] * 1.1,
            'cost': recommendation['estimated_cost'] * 0.9,  # Lower cost with better planning
            'duration': sum(move['estimated_duration'] for move in recommendation['moves']) * 1.3
        })
        
        # Alternative 3: Weather-adaptive
        if analysis['weather_impact'] < 0.8:
            alternatives.append({
                'description': 'Weather-adaptive - postpone non-critical moves',
                'moves': len(recommendation['moves']) // 3,
                'completion': recommendation['expected_completion'] * 0.4,
                'cost': recommendation['estimated_cost'] * 0.7,
                'duration': sum(move['estimated_duration'] for move in recommendation['moves'][:len(recommendation['moves'])//3])
            })
        
        # Expected outcomes
        expected_outcomes = {
            'vessel_delay_reduction': recommendation['priority_score'] * 10,  # minutes
            'yard_utilization_improvement': recommendation['expected_completion'] * 15,  # percentage
            'equipment_efficiency': len(recommendation['moves']) * 2,  # percentage
            'cost_per_move': recommendation['estimated_cost'] / max(1, len(recommendation['moves']))
        }
        
        return AIExplanation(
            primary_reasoning=primary_reasoning,
            secondary_reasoning=secondary_reasoning,
            confidence_level=confidence,
            alternatives=alternatives,
            risks=template['risks'] + analysis['risk_factors'],
            data_sources=['yard_sensors', 'equipment_telemetry', 'weather_service', 'vessel_schedule'],
            uncertainty_factors=['demand_fluctuations', 'equipment_reliability', 'weather_forecast_accuracy'],
            expected_outcomes=expected_outcomes
        )
```

```python
class HumanOperator:
    """Human operator with domain expertise and decision-making capabilities"""
    
    def __init__(self, name: str, experience_years: int, expertise_areas: List[str]):
        self.name = name
        self.experience_years = experience_years
        self.expertise_areas = expertise_areas
        self.decision_history = []
        self.insights_database = []
        self.confidence_calibration = 0.85  # How well human calibrates confidence
        
        # Human characteristics
        self.risk_tolerance = 0.6  # 0.0 = risk-averse, 1.0 = risk-seeking
        self.customer_sensitivity = 0.8  # How much they prioritize customers
        self.safety_consciousness = 0.9  # How much they prioritize safety
        self.cost_sensitivity = 0.7  # How much they consider costs
    
    def evaluate_ai_recommendation(self, recommendation: Dict, explanation: AIExplanation,
                                  context: OperationalContext, 
                                  customer_constraints: List[CustomerConstraint]) -> Dict:
        """Evaluate AI recommendation and provide human feedback"""
        
        evaluation = {
            'feedback_type': '',
            'modifications': {},
            'reasoning': '',
            'confidence': 0.0,
            'insights': [],
            'concerns': [],
            'acceptance_threshold': 0.7
        }
        
        # Human evaluation factors
        human_factors = self._evaluate_human_factors(recommendation, explanation, context)
        
        # Check for customer constraints
        customer_issues = self._check_customer_constraints(recommendation, customer_constraints)
        
        # Check for safety concerns
        safety_issues = self._check_safety_concerns(recommendation, context)
        
        # Check for operational practicality
        practicality_issues = self._check_operational_practicality(recommendation, context)
        
        # Combine all concerns
        all_concerns = customer_issues + safety_issues + practicality_issues
        
        # Make decision
        if len(all_concerns) == 0 and explanation.confidence_level >= evaluation['acceptance_threshold']:
            # Accept AI recommendation
            evaluation['feedback_type'] = 'accept'
            evaluation['confidence'] = explanation.confidence_level
            evaluation['reasoning'] = 'AI recommendation aligns with operational requirements and constraints'
        
        elif len(all_concerns) <= 2 and explanation.confidence_level >= 0.6:
            # Modify AI recommendation
            evaluation['feedback_type'] = 'modify'
            evaluation['modifications'] = self._generate_modifications(recommendation, all_concerns)
            evaluation['confidence'] = explanation.confidence_level * 0.9  # Slightly lower confidence
            evaluation['reasoning'] = f'Modified to address: {", ".join(all_concerns)}'
            evaluation['insights'] = self._extract_insights(all_concerns, context)
        
        else:
            # Request alternatives or reject
            if explanation.alternatives:
                evaluation['feedback_type'] = 'request_alternatives'
                evaluation['confidence'] = 0.5
                evaluation['reasoning'] = 'Current recommendation has significant concerns, requesting alternatives'
            else:
                evaluation['feedback_type'] = 'reject'
                evaluation['confidence'] = 0.3
                evaluation['reasoning'] = f'Recommendation rejected due to: {", ".join(all_concerns)}'
        
        evaluation['concerns'] = all_concerns
        
        return evaluation
    
    def _evaluate_human_factors(self, recommendation: Dict, explanation: AIExplanation, 
                               context: OperationalContext) -> Dict:
        """Evaluate recommendation based on human factors"""
        
        factors = {
            'risk_assessment': 0.0,
            'customer_impact': 0.0,
            'safety_assessment': 0.0,
            'cost_assessment': 0.0,
            'practicality': 0.0
        }
        
        # Risk assessment
        if explanation.risk_level == 'high':
            factors['risk_assessment'] = 0.3
        elif explanation.risk_level == 'medium':
            factors['risk_assessment'] = 0.6
        else:
            factors['risk_assessment'] = 0.9
        
        # Customer impact
        if 'customer_service' in explanation.primary_reasoning:
            factors['customer_impact'] = 0.9
        else:
            factors['customer_impact'] = 0.7
        
        # Safety assessment
        weather_risk = any('weather' in risk.lower() for risk in explanation.risks)
        equipment_risk = any('equipment' in risk.lower() for risk in explanation.risks)
        
        if weather_risk or equipment_risk:
            factors['safety_assessment'] = 0.6
        else:
            factors['safety_assessment'] = 0.9
        
        # Cost assessment
        if recommendation['estimated_cost'] > 10000:  # High cost
            factors['cost_assessment'] = 0.6
        else:
            factors['cost_assessment'] = 0.8
        
        # Practicality
        if recommendation['time_window'] == '3_hours':
            factors['practicality'] = 0.8
        else:
            factors['practicality'] = 0.9
        
        return factors
    
    def _check_customer_constraints(self, recommendation: Dict, 
                                   customer_constraints: List[CustomerConstraint]) -> List[str]:
        """Check for customer constraint violations"""
        
        concerns = []
        
        for constraint in customer_constraints:
            if constraint.priority_level == 'premium':
                # Check if moves affect premium customer containers
                for move in recommendation['moves']:
                    if move['source_block'] in [1, 2]:  # Assume premium containers in blocks 1-2
                        concerns.append(f'Premium customer {constraint.customer_name} containers affected')
                        break
            
            # Check time sensitivity
            if constraint.time_sensitivity > 0.8 and recommendation['time_window'] == '3_hours':
                concerns.append(f'Time-sensitive customer {constraint.customer_name} may experience delays')
        
        return concerns
    
    def _check_safety_concerns(self, recommendation: Dict, context: OperationalContext) -> List[str]:
        """Check for safety concerns"""
        
        concerns = []
        
        # Weather safety
        weather = context.weather_conditions
        if weather.get('wind_speed', 0) > 25:  # High wind
            if recommendation['time_window'] == '3_hours':
                concerns.append('High wind conditions - recommend reducing operation time')
        
        if weather.get('precipitation', 0) > 10:  # Heavy rain
            concerns.append('Heavy rain - increased safety risks for equipment operations')
        
        # Equipment safety
        if len(context.equipment_status) < len(recommendation.get('equipment_allocation', {})):
            concerns.append('Insufficient available equipment for safe operations')
        
        # Staff safety
        total_staff = sum(context.staff_availability.values())
        if total_staff < 10 and len(recommendation['moves']) > 5:
            concerns.append('Limited staff availability for extensive operations')
        
        return concerns
    
    def _check_operational_practicality(self, recommendation: Dict, context: OperationalContext) -> List[str]:
        """Check for operational practicality issues"""
        
        concerns = []
        
        # Gate traffic
        if context.gate_traffic_level == 'high' and len(recommendation['moves']) > 8:
            concerns.append('High gate traffic may interfere with housekeeping operations')
        
        # Maintenance activities
        if context.maintenance_activities and len(recommendation['moves']) > 6:
            concerns.append('Maintenance activities may limit equipment availability')
        
        # Time window practicality
        total_duration = sum(move['estimated_duration'] for move in recommendation['moves'])
        if recommendation['time_window'] == '3_hours' and total_duration > 150:  # More than 2.5 hours
            concerns.append('Operation duration may exceed available time window')
        
        return concerns
    
    def _generate_modifications(self, recommendation: Dict, concerns: List[str]) -> Dict:
        """Generate modifications to address concerns"""
        
        modifications = copy.deepcopy(recommendation)
        
        # Address premium customer concerns
        if any('premium customer' in concern.lower() for concern in concerns):
            # Remove moves affecting premium customers
            modifications['moves'] = [
                move for move in modifications['moves'] 
                if move['source_block'] not in [1, 2]  # Assume premium blocks
            ]
            modifications['reasoning'] = 'Protected premium customer containers'
        
        # Address weather concerns
        if any('weather' in concern.lower() or 'wind' in concern.lower() for concern in concerns):
            modifications['time_window'] = '2_hours'  # Reduce time window
            modifications['moves'] = modifications['moves'][:len(modifications['moves'])//2]  # Reduce scope
            modifications['reasoning'] = 'Reduced operation window due to weather concerns'
        
        # Address gate traffic concerns
        if any('gate traffic' in concern.lower() for concern in concerns):
            modifications['reasoning'] = 'Adjusted timing to avoid gate peak periods'
            # In practice, would adjust timing, but for now just note the modification
        
        # Recalculate derived values
        modifications['estimated_cost'] = sum(move['cost'] for move in modifications['moves'])
        modifications['priority_score'] *= 0.9  # Slightly lower priority
        modifications['expected_completion'] *= 0.8  # Lower completion percentage
        
        return modifications
    
    def _extract_insights(self, concerns: List[str], context: OperationalContext) -> List[HumanInsight]:
        """Extract human insights from concerns and context"""
        
        insights = []
        
        # Weather insight
        if any('weather' in concern.lower() for concern in concerns):
            insight = HumanInsight(
                insight_type='weather_adaptation',
                context='Weather forecast uncertainty in terminal operations',
                reasoning='Weather forecasts are often inaccurate, especially for wind events. Need larger safety buffers.',
                confidence=0.9,
                priority=0.8,
                applicable_conditions=['high_wind_forecast', 'changing_weather_patterns'],
                learned_patterns=['increase_weather_buffer_30%', 'prioritize_safety_over_efficiency']
            )
            insights.append(insight)
        
        # Customer insight
        if any('customer' in concern.lower() for concern in concerns):
            insight = HumanInsight(
                insight_type='customer_relationship',
                context='Premium customer service requirements',
                reasoning='Premium customers require special handling and minimal disruption. Their containers should be protected from housekeeping moves.',
                confidence=0.95,
                priority=0.9,
                applicable_conditions=['premium_customers', 'service_level_agreements'],
                learned_patterns=['protect_premium_containers', 'prioritize_customer_satisfaction']
            )
            insights.append(insight)
        
        return insights
```

```python
class HumanAICollaboration:
    """Main system for Human-AI collaborative housekeeping"""
    
    def __init__(self):
        self.explainable_ai = ExplainableAI()
        self.human_operator = HumanOperator(
            name="Sarah Chen",
            experience_years=12,
            expertise_areas=['yard_optimization', 'customer_service', 'safety_management']
        )
        
        self.collaboration_history = []
        self.learning_database = []
        self.performance_metrics = {
            'ai_only': [],
            'human_only': [],
            'collaborative': [],
            'improvement_over_ai': [],
            'improvement_over_human': [],
            'satisfaction_scores': [],
            'learning_captured': []
        }
    
    def simulate_collaborative_session(self, yard_state: Dict, context: OperationalContext,
                                       customer_constraints: List[CustomerConstraint]) -> CollaborativeDecision:
        """Simulate a complete human-AI collaborative decision session"""
        
        print("\n" + "="*80)
        print("HUMAN-AI COLLABORATIVE HOUSEKEEPING SESSION")
        print("="*80)
        
        # Step 1: AI generates recommendation
        print("\n=== STEP 1: AI RECOMMENDATION ===")
        ai_recommendation, ai_explanation = self.explainable_ai.generate_recommendation(
            yard_state, context, customer_constraints
        )
        
        self._display_ai_recommendation(ai_recommendation, ai_explanation)
        
        # Step 2: Human evaluation
        print("\n=== STEP 2: HUMAN EVALUATION ===")
        human_evaluation = self.human_operator.evaluate_ai_recommendation(
            ai_recommendation, ai_explanation, context, customer_constraints
        )
        
        self._display_human_evaluation(human_evaluation, ai_explanation)
        
        # Step 3: Collaborative decision
        print("\n=== STEP 3: COLLABORATIVE DECISION ===")
        final_decision = self._make_collaborative_decision(
            ai_recommendation, human_evaluation, ai_explanation
        )
        
        # Step 4: Learning capture
        print("\n=== STEP 4: LEARNING CAPTURE ===")
        learning_outcomes = self._capture_learning(
            ai_recommendation, human_evaluation, final_decision
        )
        
        # Create collaborative decision record
        collaborative_decision = CollaborativeDecision(
            timestamp=context.timestamp,
            ai_recommendation=ai_recommendation,
            ai_explanation=ai_explanation,
            human_feedback=human_evaluation,
            human_insights=human_evaluation.get('insights', []),
            final_decision=final_decision,
            collaboration_quality=self._calculate_collaboration_quality(
                ai_recommendation, human_evaluation, final_decision
            ),
            learning_outcomes=learning_outcomes
        )
        
        self._display_final_result(collaborative_decision)
        
        return collaborative_decision
    
    def _display_ai_recommendation(self, recommendation: Dict, explanation: AIExplanation):
        """Display AI recommendation and explanation"""
        
        print(f"\nAI RECOMMENDATION REPORT")
        print(f"Generated: {datetime.now().strftime('%H:%M:%S')}")
        print(f"\nPRIMARY RECOMMENDATION:")
        print(f"Execute {len(recommendation['moves'])} moves in {recommendation['time_window']} window")
        print(f"Estimated completion: {recommendation['expected_completion']:.1%}")
        print(f"Total cost: ${recommendation['estimated_cost']:,.0f}")
        
        print(f"\nREASONING FACTORS:")
        print(f"1. {explanation.primary_reasoning}")
        print(f"2. {explanation.secondary_reasoning}")
        print(f"3. Priority score: {recommendation['priority_score']:.2f}")
        
        print(f"\nCONFIDENCE LEVEL: {explanation.confidence_level:.2f} ({self._confidence_label(explanation.confidence_level)})")
        
        print(f"\nALTERNATIVE APPROACHES:")
        for i, alt in enumerate(explanation.alternatives, 1):
            print(f"Option {chr(64+i)}: {alt['description']}")
            print(f"  - Moves: {alt['moves']}, Cost: ${alt['cost']:,.0f}, Completion: {alt['completion']:.1%}")
        
        print(f"\nRISK FACTORS:")
        for risk in explanation.risks[:5]:  # Show top 5 risks
            print(f"- {risk}")
        
        print(f"\nEXPECTED OUTCOMES:")
        for key, value in explanation.expected_outcomes.items():
            if 'delay' in key:
                print(f"- {key.replace('_', ' ').title()}: {value:.1f} minutes")
            elif 'improvement' in key:
                print(f"- {key.replace('_', ' ').title()}: {value:.1f}%")
            else:
                print(f"- {key.replace('_', ' ').title()}: {value:.2f}")
    
    def _display_human_evaluation(self, evaluation: Dict, explanation: AIExplanation):
        """Display human evaluation of AI recommendation"""
        
        print(f"\nHUMAN OPERATOR EVALUATION:")
        print(f"Operator: {self.human_operator.name} ({self.human_operator.experience_years} years experience)")
        print(f"Feedback Type: {evaluation['feedback_type'].upper()}")
        print(f"Confidence: {evaluation['confidence']:.2f}")
        
        if evaluation['feedback_type'] == 'modify':
            print(f"\nHUMAN MODIFICATIONS:")
            for key, value in evaluation['modifications'].items():
                if key == 'moves':
                    print(f"- Move count: {len(value)} (was {len(explanation.alternatives[0]['moves']) if explanation.alternatives else 'N/A'})")
                elif key == 'estimated_cost':
                    print(f"- Cost: ${value:,.0f} (was ${evaluation.get('original_cost', 0):,.0f})")
                elif key == 'time_window':
                    print(f"- Time window: {value} (was 3_hours)")
                elif key == 'reasoning':
                    print(f"- Reasoning: {value}")
        
        print(f"\nHUMAN REASONING:")
        print(f"{evaluation['reasoning']}")
        
        if evaluation['concerns']:
            print(f"\nCONCERNS IDENTIFIED:")
            for concern in evaluation['concerns']:
                print(f"- {concern}")
        
        if evaluation['insights']:
            print(f"\nHUMAN INSIGHTS CAPTURED:")
            for insight in evaluation['insights']:
                print(f"- {insight.insight_type.replace('_', ' ').title()}: {insight.reasoning}")
    
    def _make_collaborative_decision(self, ai_recommendation: Dict, 
                                   human_evaluation: Dict, 
                                   ai_explanation: AIExplanation) -> Dict:
        """Make final collaborative decision"""
        
        if human_evaluation['feedback_type'] == 'accept':
            final_decision = ai_recommendation
            decision_source = 'ai_accepted'
        elif human_evaluation['feedback_type'] == 'modify':
            final_decision = human_evaluation['modifications']
            decision_source = 'human_modified'
        elif human_evaluation['feedback_type'] == 'request_alternatives':
            # Select best alternative
            best_alt = min(ai_explanation.alternatives, key=lambda x: x['cost'])
            final_decision = {
                'moves': [],  # Would be populated from alternative
                'time_window': '4_hours',
                'estimated_cost': best_alt['cost'],
                'priority_score': ai_recommendation['priority_score'] * 0.8,
                'expected_completion': best_alt['completion'],
                'source': 'alternative_selected'
            }
            decision_source = 'alternative_selected'
        else:
            # Reject and create conservative plan
            final_decision = {
                'moves': ai_recommendation['moves'][:2],  # Minimal moves
                'time_window': '2_hours',
                'estimated_cost': sum(move['cost'] for move in ai_recommendation['moves'][:2]),
                'priority_score': ai_recommendation['priority_score'] * 0.6,
                'expected_completion': 0.4,
                'source': 'conservative_plan'
            }
            decision_source = 'conservative_plan'
        
        final_decision['decision_source'] = decision_source
        final_decision['human_confidence'] = human_evaluation['confidence']
        final_decision['ai_confidence'] = ai_explanation.confidence_level
        
        return final_decision
    
    def _capture_learning(self, ai_recommendation: Dict, human_evaluation: Dict, 
                        final_decision: Dict) -> List[str]:
        """Capture learning from human-AI interaction"""
        
        learning_outcomes = []
        
        # Capture human insights
        if 'insights' in human_evaluation:
            for insight in human_evaluation['insights']:
                learning_outcomes.append(f"Learned: {insight.reasoning}")
                
                # Add to learning database
                learning_capture = LearningCapture(
                    interaction_id=f"interaction_{len(self.learning_database) + 1}",
                    timestamp=datetime.now(),
                    situation_description=insight.context,
                    human_modification=final_decision,
                    original_ai_recommendation=ai_recommendation,
                    outcome_better=True,  # Assume human modification was better
                    learning_extracted=insight.reasoning,
                    generalizable_pattern=True,
                    confidence_improvement=0.1
                )
                self.learning_database.append(learning_capture)
        
        # Capture decision pattern
        if human_evaluation['feedback_type'] == 'modify':
            learning_outcomes.append(f"Decision pattern: {final_decision.get('reasoning', 'Unknown')}")
        
        # Capture confidence calibration
        human_confidence = human_evaluation['confidence']
        ai_confidence = ai_recommendation.get('priority_score', 0.5)
        
        if abs(human_confidence - ai_confidence) > 0.2:
            learning_outcomes.append(f"Confidence calibration needed: Human {human_confidence:.2f} vs AI {ai_confidence:.2f}")
        
        return learning_outcomes
    
    def _calculate_collaboration_quality(self, ai_recommendation: Dict, 
                                         human_evaluation: Dict, 
                                         final_decision: Dict) -> float:
        """Calculate quality of human-AI collaboration"""
        
        quality_score = 0.5  # Base score
        
        # Factor 1: Agreement level
        if human_evaluation['feedback_type'] == 'accept':
            quality_score += 0.2
        elif human_evaluation['feedback_type'] == 'modify':
            quality_score += 0.15  # Good collaboration with modifications
        
        # Factor 2: Insight quality
        if 'insights' in human_evaluation and len(human_evaluation['insights']) > 0:
            quality_score += 0.15
        
        # Factor 3: Reasoning quality
        if human_evaluation.get('reasoning') and len(human_evaluation['reasoning']) > 50:
            quality_score += 0.1
        
        return min(1.0, quality_score)
    
    def _display_final_result(self, decision: CollaborativeDecision):
        """Display final collaborative decision result"""
        
        print(f"\nCOLLABORATIVE DECISION: {decision.final_decision.get('decision_source', 'unknown').upper()}")
        print(f"Moves to execute: {len(decision.final_decision.get('moves', []))}")
        print(f"Estimated cost: ${decision.final_decision.get('estimated_cost', 0):,.0f}")
        print(f"Collaboration quality: {decision.collaboration_quality:.1%}")
        
        print(f"\nEXECUTION COMPLETE:")
        print(f"Moves executed: {len(decision.final_decision.get('moves', []))}")
        print(f"Cost: ${decision.final_decision.get('estimated_cost', 0):,.0f}")
        print(f"Collaboration quality: {decision.collaboration_quality:.1%}")
        
        if decision.learning_outcomes:
            print(f"\nLEARNING CAPTURE:")
            for outcome in decision.learning_outcomes:
                print(f"- {outcome}")
        
        # Store collaboration history
        self.collaboration_history.append(decision)
    
    def _confidence_label(self, confidence: float) -> str:
        """Convert confidence score to label"""
        if confidence >= 0.8:
            return "High"
        elif confidence >= 0.6:
            return "Medium"
        elif confidence >= 0.4:
            return "Low"
        else:
            return "Very Low"
    
    def analyze_collaboration_performance(self) -> Dict:
        """Analyze overall collaboration performance"""
        
        if not self.collaboration_history:
            return {'message': 'No collaboration history available'}
        
        analysis = {
            'total_sessions': len(self.collaboration_history),
            'avg_collaboration_quality': np.mean([d.collaboration_quality for d in self.collaboration_history]),
            'decision_distribution': {},
            'learning_captured': len(self.learning_database),
            'improvement_trends': {}
        }
        
        # Decision distribution
        decision_sources = [d.final_decision.get('decision_source', 'unknown') for d in self.collaboration_history]
        for source in set(decision_sources):
            analysis['decision_distribution'][source] = decision_sources.count(source) / len(decision_sources)
        
        return analysis
```
