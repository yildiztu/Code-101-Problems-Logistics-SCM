# 46. The Competitive Port Pricing Problem

## Tier 10: The Proactive Demand & Market Shaping

### Goal
Transform from reactive pricing to proactive market influence, where ports don't just respond to market conditions but actively shape demand patterns, influence customer behavior, and create new market opportunities through strategic pricing interventions.

### Key Assumptions
- Ports can influence customer behavior through strategic pricing
- Market conditions can be shaped through coordinated interventions
- Long-term market positioning requires proactive strategic actions
- Customer segments respond differently to strategic incentives

### Approach (Step-by-Step)
1. Develop market influence models and demand elasticity analysis
2. Create customer segmentation and behavioral profiling
3. Implement strategic pricing interventions and incentive programs
4. Build market trend forecasting and opportunity identification
5. Develop partnership programs and ecosystem development
6. Create competitive positioning and market leadership strategies

### What to Look for in the Results
- Proactive market interventions that shift demand patterns
- Strategic pricing that creates new market opportunities
- Customer behavior changes driven by port initiatives
- Long-term competitive advantage through market shaping

### Concrete Example (from the source)
Through proactive market shaping, a port implemented a three-tier strategy: (1) introduced volume discounts for emerging trade lanes, (2) created premium services for high-value customers, and (3) developed partnership programs with logistics providers. This resulted in 25% growth in new market segments, 18% increase in customer loyalty, and 32% improvement in market position while maintaining ethical standards and profitability.

```python
# Import required libraries for Proactive Demand & Market Shaping
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass, field
from typing import List, Tuple, Dict, Optional, Any, Set, Union
from collections import defaultdict, deque
import warnings
import random
import time
from enum import Enum
warnings.filterwarnings('ignore')

# Set plotting style for professional visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
@dataclass
class Port:
    """Port with capacity and cost structure"""
    id: int
    name: str
    capacity: float  # Maximum TEU capacity
    costs: List[float]  # Cost per shipping line
    strategic_position: str  # 'leader', 'challenger', 'follower'
    
@dataclass
class ShippingLine:
    """Shipping line with demand characteristics"""
    id: int
    name: str
    demand: float  # TEU demand
    priority: int  # Priority level (1=high, 2=medium, 3=low)
    segment: str  # 'premium', 'standard', 'budget'
    price_elasticity: float  # Price elasticity of demand
    growth_potential: float  # Growth potential (0-1)
    loyalty_score: float  # Customer loyalty (0-1)
    
@dataclass
class MarketParameters:
    """Market-wide parameters"""
    price_sensitivity: float  # Beta parameter for logit model
    min_price: float  # Minimum price
    max_price: float  # Maximum price
    market_growth_rate: float  # Annual market growth rate
    competitive_intensity: float  # Level of competition (0-1)

class MarketInterventionType(Enum):
    """Types of market interventions"""
    VOLUME_DISCOUNT = "volume_discount"
    PREMIUM_SERVICE = "premium_service"
    PARTNERSHIP_PROGRAM = "partnership_program"
    LOYALTY_INCENTIVE = "loyalty_incentive"
    NEW_MARKET_DEVELOPMENT = "new_market_development"
    COMPETITIVE_POSITIONING = "competitive_positioning"
    DEMAND_STIMULATION = "demand_stimulation"
    BEHAVIORAL_NUDGE = "behavioral_nudge"

@dataclass
class MarketIntervention:
    """Strategic market intervention"""
    intervention_type: MarketInterventionType
    target_segments: List[str]
    pricing_strategy: str
    expected_impact: Dict[str, float]
    duration_weeks: int
    investment_required: float
    success_metrics: List[str]
    risk_factors: List[str]

@dataclass
class CustomerBehavior:
    """Customer behavior patterns"""
    segment: str
    price_sensitivity: float
    service_preference: float
    loyalty_drivers: List[str]
    switching_costs: float
    growth_indicators: List[str]

@dataclass
class MarketOpportunity:
    """Identified market opportunity"""
    opportunity_type: str
    target_customers: List[str]
    market_size: float
    growth_rate: float
    profit_potential: float
    competitive_advantage: str
    required_capabilities: List[str]
    time_to_market: int  # weeks
    confidence_level: float
```

```python
class MarketShapingEngine:
    """Advanced market shaping and demand influence system"""
    
    def __init__(self, ports: List[Port], shipping_lines: List[ShippingLine], 
                 market_params: MarketParameters):
        self.ports = ports
        self.shipping_lines = shipping_lines
        self.market_params = market_params
        
        self.num_ports = len(ports)
        self.num_shipping_lines = len(shipping_lines)
        
        # Market intelligence
        self.market_trends = defaultdict(list)
        self.customer_behaviors = {}
        self.competitive_landscape = {}
        
        # Strategic interventions
        self.active_interventions = []
        self.intervention_history = []
        self.intervention_performance = defaultdict(list)
        
        # Market opportunities
        self.identified_opportunities = []
        self.opportunity_pipeline = []
        
        # Performance metrics
        self.market_shaping_metrics = {
            'demand_shift': [],
            'behavior_change': [],
            'market_growth': [],
            'competitive_position': [],
            'customer_loyalty': [],
            'new_market_penetration': []
        }
        
        # Initialize customer behaviors
        self._initialize_customer_behaviors()
        
        # Initialize competitive landscape
        self._initialize_competitive_landscape()
    
    def _initialize_customer_behaviors(self):
        """Initialize customer behavior profiles"""
        
        behavior_profiles = {
            'premium': CustomerBehavior(
                segment='premium',
                price_sensitivity=0.3,  # Low price sensitivity
                service_preference=0.9,  # High service preference
                loyalty_drivers=['service_quality', 'reliability', 'exclusive_access'],
                switching_costs=0.8,  # High switching costs
                growth_indicators=['willingness_to_pay_premium', 'demand_for_value_added']
            ),
            'standard': CustomerBehavior(
                segment='standard',
                price_sensitivity=0.6,  # Medium price sensitivity
                service_preference=0.6,  # Medium service preference
                loyalty_drivers=['price_stability', 'service_consistency', 'relationship'],
                switching_costs=0.5,  # Medium switching costs
                growth_indicators=['volume_discounts', 'service_improvements']
            ),
            'budget': CustomerBehavior(
                segment='budget',
                price_sensitivity=0.9,  # High price sensitivity
                service_preference=0.3,  # Low service preference
                loyalty_drivers=['low_price', 'basic_reliability'],
                switching_costs=0.2,  # Low switching costs
                growth_indicators=['price_promotions', 'cost_efficiency']
            )
        }
        
        for shipping_line in self.shipping_lines:
            self.customer_behaviors[shipping_line.id] = behavior_profiles[shipping_line.segment]
    
    def _initialize_competitive_landscape(self):
        """Initialize competitive landscape analysis"""
        
        self.competitive_landscape = {
            'market_leaders': [],
            'price_warriors': [],
            'service_specialists': [],
            'niche_players': [],
            'market_challengers': []
            'intensity_analysis': {
                'price_competition': 0.7,
                'service_competition': 0.6,
                'innovation_competition': 0.4,
                'partnership_competition': 0.5
            }
        }
    
    def analyze_market_opportunities(self) -> List[MarketOpportunity]:
        """Identify and analyze market opportunities"""
        
        opportunities = []
        
        # Opportunity 1: Emerging trade lanes
        emerging_lanes_opportunity = MarketOpportunity(
            opportunity_type='emerging_trade_lanes',
            target_customers=['growth_focused_lines', 'regional_expansion'],
            market_size=50000,  # TEU
            growth_rate=0.25,  # 25% annual growth
            profit_potential=0.18,  # 18% margin
            competitive_advantage='first_mover_advantage',
            required_capabilities=['flexible_infrastructure', 'customs_efficiency'],
            time_to_market=12,  # 12 weeks
            confidence_level=0.8
        )
        opportunities.append(emerging_lanes_opportunity)
        
        # Opportunity 2: Premium service segments
        premium_service_opportunity = MarketOpportunity(
            opportunity_type='premium_services',
            target_customers=['high_value_lines', 'time_sensitive_cargo'],
            market_size=30000,
            growth_rate=0.15,
            profit_potential=0.25,  # 25% margin
            competitive_advantage='service_excellence',
            required_capabilities=['premium_facilities', 'dedicated_staff'],
            time_to_market=8,
            confidence_level=0.9
        )
        opportunities.append(premium_service_opportunity)
        
        # Opportunity 3: Partnership ecosystem
        partnership_opportunity = MarketOpportunity(
            opportunity_type='partnership_ecosystem',
            target_customers=['logistics_providers', 'integrated_shippers'],
            market_size=40000,
            growth_rate=0.20,
            profit_potential=0.12,  # 12% margin (shared with partners)
            competitive_advantage='network_effects',
            required_capabilities=['digital_platform', 'partner_management'],
            time_to_market=16,
            confidence_level=0.7
        )
        opportunities.append(partnership_opportunity)
        
        self.identified_opportunities = opportunities
        return opportunities
    
    def design_strategic_intervention(self, opportunity: MarketOpportunity) -> MarketIntervention:
        """Design strategic intervention for market opportunity"""
        
        if opportunity.opportunity_type == 'emerging_trade_lanes':
            intervention = MarketIntervention(
                intervention_type=MarketInterventionType.NEW_MARKET_DEVELOPMENT,
                target_segments=['growth_focused', 'regional'],
                pricing_strategy='introductory_pricing_with_growth_incentives',
                expected_impact={
                    'demand_growth': 0.25,
                    'market_share_gain': 0.15,
                    'customer_acquisition': 0.30
                },
                duration_weeks=24,
                investment_required=500000,
                success_metrics=['volume_growth', 'new_customer_count', 'market_penetration'],
                risk_factors=['market_adoption_risk', 'competitive_response']
            )
        
        elif opportunity.opportunity_type == 'premium_services':
            intervention = MarketIntervention(
                intervention_type=MarketInterventionType.PREMIUM_SERVICE,
                target_segments=['premium', 'high_value'],
                pricing_strategy='value_based_pricing_with_service_premium',
                expected_impact={
                    'revenue_growth': 0.20,
                    'margin_improvement': 0.08,
                    'customer_loyalty': 0.18
                },
                duration_weeks=12,
                investment_required=300000,
                success_metrics=['premium_revenue', 'customer_satisfaction', 'retention_rate'],
                risk_factors=['service_delivery_risk', 'price_acceptance']
            )
        
        else:  # partnership_ecosystem
            intervention = MarketIntervention(
                intervention_type=MarketInterventionType.PARTNERSHIP_PROGRAM,
                target_segments=['integrated', 'logistics_focused'],
                pricing_strategy='revenue_sharing_with_partners',
                expected_impact={
                    'ecosystem_growth': 0.20,
                    'network_effects': 0.25,
                    'market_lock_in': 0.12
                },
                duration_weeks=48,
                investment_required=750000,
                success_metrics=['partner_revenue', 'network_size', 'integration_depth'],
                risk_factors=['partner_coordination', 'revenue_sharing_conflicts']
            )
        
        return intervention
    
    def implement_intervention(self, intervention: MarketIntervention) -> Dict:
        """Implement strategic market intervention"""
        
        print(f"\nImplementing {intervention.intervention_type.value} intervention:")
        print(f"Target segments: {', '.join(intervention.target_segments)}")
        print(f"Pricing strategy: {intervention.pricing_strategy}")
        print(f"Duration: {intervention.duration_weeks} weeks")
        print(f"Investment: ${intervention.investment_required:,.0f}")
        
        # Add to active interventions
        self.active_interventions.append(intervention)
        
        # Simulate intervention implementation
        implementation_results = self._simulate_intervention_implementation(intervention)
        
        return implementation_results
    
    def _simulate_intervention_implementation(self, intervention: MarketIntervention) -> Dict:
        """Simulate the implementation and effects of intervention"""
        
        # Initialize results
        results = {
            'intervention_type': intervention.intervention_type.value,
            'weekly_progress': [],
            'final_outcomes': {},
            'success_metrics': {},
            'roi': 0.0
        }
        
        # Simulate weekly progress
        for week in range(intervention.duration_weeks):
            weekly_result = {
                'week': week + 1,
                'demand_change': 0.0,
                'behavior_change': 0.0,
                'market_position_change': 0.0,
                'customer_loyalty_change': 0.0
            }
            
            # Calculate intervention effects based on type
            if intervention.intervention_type == MarketInterventionType.NEW_MARKET_DEVELOPMENT:
                # Gradual market development
                adoption_rate = min(1.0, week / 12.0)  # Full adoption by week 12
                weekly_result['demand_change'] = intervention.expected_impact['demand_growth'] * adoption_rate * 0.1
                weekly_result['behavior_change'] = 0.02 * adoption_rate
                
            elif intervention.intervention_type == MarketInterventionType.PREMIUM_SERVICE:
                # Faster premium service adoption
                adoption_rate = min(1.0, week / 8.0)  # Full adoption by week 8
                weekly_result['demand_change'] = intervention.expected_impact['revenue_growth'] * adoption_rate * 0.08
                weekly_result['customer_loyalty_change'] = intervention.expected_impact['customer_loyalty'] * adoption_rate * 0.02
                
            elif intervention.intervention_type == MarketInterventionType.PARTNERSHIP_PROGRAM:
                # Slower partnership ecosystem development
                adoption_rate = min(1.0, week / 24.0)  # Full adoption by week 24
                weekly_result['demand_change'] = intervention.expected_impact['ecosystem_growth'] * adoption_rate * 0.05
                weekly_result['market_position_change'] = intervention.expected_impact['network_effects'] * adoption_rate * 0.03
            
            results['weekly_progress'].append(weekly_result)
        
        # Calculate final outcomes
        total_demand_change = sum(w['demand_change'] for w in results['weekly_progress'])
        total_behavior_change = sum(w['behavior_change'] for w in results['weekly_progress'])
        total_position_change = sum(w['market_position_change'] for w in results['weekly_progress'])
        total_loyalty_change = sum(w['customer_loyalty_change'] for w in results['weekly_progress'])
        
        results['final_outcomes'] = {
            'total_demand_growth': total_demand_change,
            'total_behavior_change': total_behavior_change,
            'market_position_improvement': total_position_change,
            'customer_loyalty_improvement': total_loyalty_change
        }
        
        # Calculate success metrics
        results['success_metrics'] = {
            'target_achievement': self._calculate_target_achievement(intervention, results['final_outcomes']),
            'roi': self._calculate_roi(intervention, results['final_outcomes']),
            'risk_mitigation': self._assess_risk_mitigation(intervention, results['weekly_progress'])
        }
        
        return results
    
    def _calculate_target_achievement(self, intervention: MarketIntervention, outcomes: Dict) -> float:
        """Calculate achievement of intervention targets"""
        
        achievement_scores = []
        
        for target, expected_value in intervention.expected_impact.items():
            if target in outcomes:
                actual_value = outcomes[target]
                achievement = min(1.0, actual_value / expected_value) if expected_value > 0 else 0.0
                achievement_scores.append(achievement)
        
        return np.mean(achievement_scores) if achievement_scores else 0.0
    
    def _calculate_roi(self, intervention: MarketIntervention, outcomes: Dict) -> float:
        """Calculate return on investment for intervention"""
        
        # Simplified ROI calculation
        total_impact = sum(outcomes.values())
        estimated_revenue_gain = total_impact * 1000000  # Convert to monetary value
        roi = (estimated_revenue_gain - intervention.investment_required) / intervention.investment_required
        
        return roi
    
    def _assess_risk_mitigation(self, intervention: MarketIntervention, weekly_progress: List[Dict]) -> float:
        """Assess how well risks were mitigated"""
        
        # Simplified risk assessment based on progress stability
        if len(weekly_progress) < 2:
            return 0.5
        
        # Check for consistent progress (low volatility indicates good risk mitigation)
        demand_changes = [w['demand_change'] for w in weekly_progress]
        volatility = np.std(demand_changes) / (np.mean(demand_changes) + 1e-6)
        
        # Lower volatility = better risk mitigation
        risk_mitigation_score = max(0.0, 1.0 - volatility)
        
        return risk_mitigation_score
    
    def calculate_market_shaping_impact(self) -> Dict:
        """Calculate overall market shaping impact"""
        
        # Aggregate results from all interventions
        total_demand_shift = 0.0
        total_behavior_change = 0.0
        total_market_growth = 0.0
        total_position_improvement = 0.0
        total_loyalty_improvement = 0.0
        
        for intervention in self.active_interventions:
            if hasattr(intervention, 'implementation_results'):
                results = intervention.implementation_results
                outcomes = results.get('final_outcomes', {})
                
                total_demand_shift += outcomes.get('total_demand_growth', 0.0)
                total_behavior_change += outcomes.get('total_behavior_change', 0.0)
                total_position_improvement += outcomes.get('market_position_improvement', 0.0)
                total_loyalty_improvement += outcomes.get('customer_loyalty_improvement', 0.0)
        
        # Calculate market growth (from demand shift and new market penetration)
        total_market_growth = total_demand_shift * 1.2  # Market multiplier effect
        
        impact_summary = {
            'demand_shift_percentage': total_demand_shift * 100,
            'behavior_change_score': total_behavior_change,
            'market_growth_percentage': total_market_growth * 100,
            'competitive_position_gain': total_position_improvement * 100,
            'customer_loyalty_improvement': total_loyalty_improvement * 100,
            'overall_market_shaping_score': (total_demand_shift + total_behavior_change + 
                                           total_market_growth + total_position_improvement + 
                                           total_loyalty_improvement) / 5.0
        }
        
        return impact_summary
```

```python
# Create the concrete example from the source text
def create_concrete_example():
    """Create the example with 3 ports and 3 shipping lines"""
    
    # Define ports with strategic positions
    ports = [
        Port(id=0, name="Port A", capacity=20000, costs=[80, 85, 82], strategic_position='leader'),
        Port(id=1, name="Port B", capacity=18000, costs=[85, 78, 88], strategic_position='challenger'),
        Port(id=2, name="Port C", capacity=22000, costs=[75, 90, 80], strategic_position='follower')
    ]
    
    # Define shipping lines with diverse characteristics
    shipping_lines = [
        ShippingLine(id=0, name="Global Shipping Line", demand=12000, priority=1, 
                   segment='premium', price_elasticity=-0.3, growth_potential=0.8, loyalty_score=0.9),
        ShippingLine(id=1, name="Pacific Trade Line", demand=8000, priority=2, 
                   segment='standard', price_elasticity=-0.6, growth_potential=0.6, loyalty_score=0.7),
        ShippingLine(id=2, name="Regional Container Line", demand=5000, priority=3, 
                   segment='budget', price_elasticity=-0.9, growth_potential=0.4, loyalty_score=0.5)
    ]
    
    # Market parameters with growth and competition
    market_params = MarketParameters(
        price_sensitivity=0.01,  # β = 0.01
        min_price=80,
        max_price=150,
        market_growth_rate=0.12,  # 12% annual growth
        competitive_intensity=0.7  # High competition
    )
    
    return ports, shipping_lines, market_params

# Create the problem instance
ports, shipping_lines, market_params = create_concrete_example()

# Create the market shaping engine
market_shaper = MarketShapingEngine(ports, shipping_lines, market_params)

print("Proactive Demand & Market Shaping Configuration:")
print(f"Number of ports: {len(ports)}")
print(f"Number of shipping lines: {len(shipping_lines)}")
print(f"Total market demand: {sum(sl.demand for sl in shipping_lines):,} TEU")
print(f"Total port capacity: {sum(port.capacity for port in ports):,} TEU")
print(f"Market growth rate: {market_params.market_growth_rate:.1%}")
print(f"Competitive intensity: {market_params.competitive_intensity:.1%}")
print(f"\nPort Strategic Positions:")
for port in ports:
    print(f"- {port.name}: {port.strategic_position.title()}")
print(f"\nShipping Line Segments:")
for sl in shipping_lines:
    print(f"- {sl.name}: {sl.segment.title()}, Growth Potential {sl.growth_potential:.1%}, Loyalty {sl.loyalty_score:.1%}")
print(f"\nCustomer Behavior Profiles:")
for sl_id, behavior in market_shaper.customer_behaviors.items():
    sl = next(sl for sl in shipping_lines if sl.id == sl_id)
    print(f"- {sl.name}: Price Sensitivity {behavior.price_sensitivity:.1f}, Service Preference {behavior.service_preference:.1f}")
```

```python
# Strategic Intervention Design and Implementation
def implement_strategic_interventions():
    """Design and implement strategic interventions"""
    
    print("\n" + "="*80)
    print("STRATEGIC INTERVENTION DESIGN & IMPLEMENTATION")
    print("="*80)
    
    intervention_results = []
    
    for i, opportunity in enumerate(opportunities, 1):
        print(f"\n{'='*60}")
        print(f"INTERVENTION {i}: {opportunity.opportunity_type.replace('_', ' ').title()}")
        print(f"{'='*60}")
        
        # Design intervention
        intervention = market_shaper.design_strategic_intervention(opportunity)
        
        # Implement intervention
        results = market_shaper.implement_intervention(intervention)
        
        # Store results
        intervention.implementation_results = results
        intervention_results.append(results)
        
        # Display results
        print(f"\nImplementation Results:")
        print(f"Target Achievement: {results['success_metrics']['target_achievement']:.1%}")
        print(f"ROI: {results['success_metrics']['roi']:.1%}")
        print(f"Risk Mitigation: {results['success_metrics']['risk_mitigation']:.1%}")
        
        print(f"\nFinal Outcomes:")
        for outcome, value in results['final_outcomes'].items():
            print(f"- {outcome.replace('_', ' ').title()}: {value:.2%}")
    
    return intervention_results

# Implement strategic interventions
intervention_results = implement_strategic_interventions()
```

```python
# Comprehensive Visualization of Market Shaping Results
def visualize_market_shaping():
    """Create comprehensive visualizations of market shaping results"""
    
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    fig.suptitle('Proactive Demand & Market Shaping Analysis', fontsize=16, fontweight='bold')
    
    # Plot 1: Intervention Progress Over Time
    ax1 = axes[0, 0]
    
    for i, results in enumerate(intervention_results):
        weeks = [w['week'] for w in results['weekly_progress']]
        demand_changes = [w['demand_change'] * 100 for w in results['weekly_progress']]
        ax1.plot(weeks, demand_changes, label=f"Intervention {i+1}", linewidth=2)
    
    ax1.set_title('Demand Change Over Time by Intervention')
    ax1.set_xlabel('Week')
    ax1.set_ylabel('Demand Change (%)')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Plot 2: Success Metrics Comparison
    ax2 = axes[0, 1]
    
    interventions = ['New Market', 'Premium Service', 'Partnership']
    target_achievement = [r['success_metrics']['target_achievement'] for r in intervention_results]
    roi_values = [r['success_metrics']['roi'] for r in intervention_results]
    risk_mitigation = [r['success_metrics']['risk_mitigation'] for r in intervention_results]
    
    x = np.arange(len(interventions))
    width = 0.25
    
    bars1 = ax2.bar(x - width, target_achievement, width, label='Target Achievement', alpha=0.7)
    bars2 = ax2.bar(x, roi_values, width, label='ROI', alpha=0.7)
    bars3 = ax2.bar(x + width, risk_mitigation, width, label='Risk Mitigation', alpha=0.7)
    
    ax2.set_title('Intervention Success Metrics')
    ax2.set_ylabel('Score / ROI')
    ax2.set_xticks(x)
    ax2.set_xticklabels(interventions)
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # Plot 3: Market Opportunity Assessment
    ax3 = axes[0, 2]
    
    opportunity_names = [opp.opportunity_type.replace('_', ' ').title() for opp in opportunities]
    market_sizes = [opp.market_size / 1000 for opp in opportunities]  # Convert to thousands
    growth_rates = [opp.growth_rate * 100 for opp in opportunities]
    confidence_levels = [opp.confidence_level * 100 for opp in opportunities]
    
    x = np.arange(len(opportunity_names))
    width = 0.25
    
    bars1 = ax3.bar(x - width, market_sizes, width, label='Market Size (k TEU)', alpha=0.7)
    bars2 = ax3.bar(x, growth_rates, width, label='Growth Rate (%)', alpha=0.7)
    bars3 = ax3.bar(x + width, confidence_levels, width, label='Confidence (%)', alpha=0.7)
    
    ax3.set_title('Market Opportunity Assessment')
    ax3.set_ylabel('Value')
    ax3.set_xticks(x)
    ax3.set_xticklabels([name[:15] for name in opportunity_names], rotation=45)
    ax3.legend()
    ax3.grid(True, alpha=0.3)
    
    # Plot 4: Overall Market Shaping Impact
    ax4 = axes[1, 0]
    
    impact_metrics = ['Demand\nShift', 'Market\nGrowth', 'Position\nGain', 'Loyalty\nImprovement']
    impact_values = [
        impact_summary['demand_shift_percentage'],
        impact_summary['market_growth_percentage'],
        impact_summary['competitive_position_gain'],
        impact_summary['customer_loyalty_improvement']
    ]
    
    bars = ax4.bar(impact_metrics, impact_values, color='green', alpha=0.7)
    ax4.set_title('Overall Market Shaping Impact')
    ax4.set_ylabel('Impact (%)')
    ax4.grid(True, alpha=0.3)
    
    # Add value labels
    for bar, value in zip(bars, impact_values):
        height = bar.get_height()
        ax4.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                f'{value:.1f}%', ha='center', va='bottom')
    
    # Plot 5: Customer Behavior Changes
    ax5 = axes[1, 1]
    
    # Aggregate behavior changes across interventions
    total_behavior_changes = []
    for results in intervention_results:
        behavior_changes = [w['behavior_change'] for w in results['weekly_progress']]
        total_behavior_changes.extend(behavior_changes)
    
    # Calculate cumulative behavior change
    cumulative_behavior_change = np.cumsum(total_behavior_changes)
    weeks = range(1, len(cumulative_behavior_change) + 1)
    
    ax5.plot(weeks, cumulative_behavior_change, 'purple', linewidth=2)
    ax5.set_title('Cumulative Customer Behavior Change')
    ax5.set_xlabel('Week')
    ax5.set_ylabel('Behavior Change Score')
    ax5.grid(True, alpha=0.3)
    
    # Plot 6: Market Shaping Strategy Comparison
    ax6 = axes[1, 2]
    
    # Compare with source example targets
    source_targets = {
        'New Segment\nGrowth': 25,
        'Customer\nLoyalty': 18,
        'Market\nPosition': 32
    }
    
    achieved_values = {
        'New Segment\nGrowth': impact_summary['market_growth_percentage'],
        'Customer\nLoyalty': impact_summary['customer_loyalty_improvement'],
        'Market\nPosition': impact_summary['competitive_position_gain']
    }
    
    metrics = list(source_targets.keys())
    source_vals = list(source_targets.values())
    achieved_vals = list(achieved_values.values())
    
    x = np.arange(len(metrics))
    width = 0.35
    
    bars1 = ax6.bar(x - width/2, source_vals, width, label='Source Target', alpha=0.7)
    bars2 = ax6.bar(x + width/2, achieved_vals, width, label='Achieved', alpha=0.7)
    
    ax6.set_title('Achievement vs Source Targets')
    ax6.set_ylabel('Percentage (%)')
    ax6.set_xticks(x)
    ax6.set_xticklabels(metrics)
    ax6.legend()
    ax6.grid(True, alpha=0.3)
    
    # Add achievement percentages
    for i, (source, achieved) in enumerate(zip(source_vals, achieved_vals)):
        achievement_pct = (achieved / source) * 100
        ax6.text(i, max(source, achieved) + 1, f'{achievement_pct:.0f}%', 
                ha='center', va='bottom', fontweight='bold')
    
    plt.tight_layout()
    plt.show()
    
    return fig

# Create visualizations
visualization_fig = visualize_market_shaping()
```
