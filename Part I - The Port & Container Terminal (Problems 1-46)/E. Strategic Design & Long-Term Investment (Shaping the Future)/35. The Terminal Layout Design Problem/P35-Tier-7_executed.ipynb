# 35. The Terminal Layout Design Problem

## Tier 7: The Human-AI Symbiotic Partnership (The Centaur Model)

### Goal
Create a human-AI collaborative system that combines computational optimization power with human expertise and contextual understanding to achieve superior terminal layout design decisions through explainable AI and interactive refinement.

### Key assumptions
- Human experts possess domain knowledge that may not be captured in mathematical models
- AI can process vast solution spaces and identify non-obvious optimization opportunities
- Explainable AI can provide transparent reasoning for layout recommendations
- Human-AI collaboration can balance quantitative metrics with qualitative factors

### Approach (step-by-step)
1. Design explainable AI system with transparent decision reasoning
2. Implement interactive human-AI collaboration framework
3. Create multi-criteria evaluation balancing quantitative and qualitative factors
4. Develop learning system that improves from human feedback and decisions
5. Demonstrate collaborative refinement process with real-world constraints

### What to look for in the results
- Explainable AI recommendations with clear reasoning and feature importance
- Interactive refinement process showing human expertise integration
- Multi-criteria analysis balancing throughput, safety, and regulatory compliance
- Learning system improvement from human-AI interactions
- Superior outcomes compared to purely AI-optimized solutions

### Concrete example (from the source)
We'll implement the Port of Hamburg Container Terminal Altenwerder example from the source material, demonstrating how the AI identified an optimal configuration with 18% throughput improvement, but human experts recognized safety issues with narrow aisles in high-wind conditions, leading to collaborative refinement that achieved 14% improvement while maintaining safety standards and stakeholder satisfaction.

```python
from typing import Tuple, List, Dict, Set

# Import required packages for human-AI collaborative system
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass, field
from typing import List, Dict, Tuple, Optional, Any
import random
import time
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
random.seed(42)

# Set plotting style
plt.style.use('default')
sns.set_palette("husl")
```

```python
try:
    @dataclass
    class LayoutSolution:
        """Represents a terminal layout solution with comprehensive attributes"""
        facility_assignments: Dict[str, Tuple[float, float]]  # facility -> (x, y)
        quantitative_metrics: Dict[str, float]  # throughput, cost, utilization, etc.
        qualitative_factors: Dict[str, float]  # safety, flexibility, environmental, etc.
        ai_confidence: float  # AI confidence in this solution
        human_acceptance: Optional[float] = None  # Human expert acceptance score
        explanation: Optional[Dict[str, Any]] = None  # AI explanation
        
    @dataclass
    class HumanExpert:
        """Represents a human domain expert"""
        name: str
        expertise_areas: List[str]  # 'safety', 'operations', 'environmental', 'regulatory'
        experience_years: int
        decision_weights: Dict[str, float]  # weights for different decision factors
        
    @dataclass
    class AIRecommendation:
        """Represents an AI recommendation with explanations"""
        solution: LayoutSolution
        primary_objective: str
        feature_importance: Dict[str, float]
        reasoning: List[str]
        alternatives: List[LayoutSolution]
        confidence_score: float
        
    @dataclass
    class CollaborationSession:
        """Tracks a human-AI collaboration session"""
        session_id: str
     human_expert: HumanExpert
        ai_recommendations: List[AIRecommendation]
        human_feedback: List[Dict[str, Any]]
        final_solution: Optional[LayoutSolution] = None
        collaboration_metrics: Dict[str, float] = field(default_factory=dict)
    
    print("Human-AI collaboration data structures defined successfully")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    class ExplainableAIOptimizer:
        """Explainable AI system for terminal layout optimization"""
        
        def __init__(self):
            self.facility_types = ['berth', 'yard_block', 'gate', 'maintenance', 'rail']
            self.constraint_types = ['safety', 'regulatory', 'environmental', 'operational']
            
            # AI model for performance prediction
            self.performance_model = None
            self.scaler = StandardScaler()
            self.feature_names = []
            
            # Historical data for learning
            self.historical_solutions = []
            self.human_preferences = {}
            
            # Explanation templates
            self.explanation_templates = {
                'throughput': 'This layout maximizes throughput by minimizing transport distances between {facilities}.',
                'safety': 'Safety is enhanced by maintaining {distance}m separation between {facilities}.',
                'cost': 'Cost efficiency is achieved through optimal facility placement reducing {metric} by {percentage}%.',
                'flexibility': 'Future expansion capability is preserved by allocating buffer zones around {facilities}.'
            }
        
        def generate_ai_solutions(self, num_solutions: int = 10) -> List[LayoutSolution]:
            """Generate AI-optimized layout solutions"""
            solutions = []
            
            for i in range(num_solutions):
                # Generate facility assignments
                facility_assignments = self._generate_facility_layout()
                
                # Calculate quantitative metrics
                quantitative_metrics = self._calculate_quantitative_metrics(facility_assignments)
                
                # Calculate qualitative factors
                qualitative_factors = self._calculate_qualitative_factors(facility_assignments)
                
                # AI confidence based on optimization quality
                ai_confidence = self._calculate_ai_confidence(quantitative_metrics, qualitative_factors)
                
                solution = LayoutSolution(
                    facility_assignments=facility_assignments,
                    quantitative_metrics=quantitative_metrics,
                    qualitative_factors=qualitative_factors,
                    ai_confidence=ai_confidence
                )
                
                solutions.append(solution)
            
            return solutions
        
        def _generate_facility_layout(self) -> Dict[str, Tuple[float, float]]:
            """Generate optimized facility layout"""
            # Terminal area: 1000m x 800m
            terminal_width = 1000
            terminal_length = 800
            
            assignments = {}
            
            # Strategic placement based on functional requirements
            # Berth facilities (waterside)
            assignments['berth_1'] = (50, 0)  # Northern berth
            assignments['berth_2'] = (450, 0)  # Central berth
            assignments['berth_3'] = (850, 0)  # Southern berth
            
            # Yard blocks (central area for optimal access)
            assignments['yard_block_1'] = (200, 300)
            assignments['yard_block_2'] = (500, 300)
            assignments['yard_block_3'] = (800, 300)
            
            # Gate facilities (landside for truck access)
            assignments['gate_1'] = (100, 700)
            assignments['gate_2'] = (500, 700)
            assignments['gate_3'] = (900, 700)
            
            # Support facilities
            assignments['maintenance'] = (500, 150)
            assignments['rail_terminal'] = (50, 500)
            assignments['admin_building'] = (500, 500)
            
            return assignments
        
        def _calculate_quantitative_metrics(self, assignments: Dict[str, Tuple[float, float]]) -> Dict[str, float]:
            """Calculate quantitative performance metrics"""
            # Extract coordinates
            berth_coords = [coord for name, coord in assignments.items() if 'berth' in name]
            yard_coords = [coord for name, coord in assignments.items() if 'yard' in name]
            gate_coords = [coord for name, coord in assignments.items() if 'gate' in name]
            
            # Calculate average transport distances
            berth_yard_distances = []
            yard_gate_distances = []
            
            for berth_coord in berth_coords:
                for yard_coord in yard_coords:
                    distance = np.sqrt((berth_coord[0] - yard_coord[0])**2 + (berth_coord[1] - yard_coord[1])**2)
                    berth_yard_distances.append(distance)
            
            for yard_coord in yard_coords:
                for gate_coord in gate_coords:
                    distance = np.sqrt((yard_coord[0] - gate_coord[0])**2 + (yard_coord[1] - gate_coord[1])**2)
                    yard_gate_distances.append(distance)
            
            avg_berth_yard = np.mean(berth_yard_distances) if berth_yard_distances else 0
            avg_yard_gate = np.mean(yard_gate_distances) if yard_gate_distances else 0
            
            # Calculate metrics
            throughput_score = 1000 / (avg_berth_yard + avg_yard_gate)  # Higher is better
            transport_efficiency = 100 / (avg_berth_yard + avg_yard_gate + 1)
            land_utilization = len(assignments) / 20  # Normalized utilization
            
            return {
                'throughput_score': throughput_score,
                'transport_efficiency': transport_efficiency,
                'avg_berth_yard_distance': avg_berth_yard,
                'avg_yard_gate_distance': avg_yard_gate,
                'land_utilization': land_utilization,
                'estimated_cost': avg_berth_yard * 10 + avg_yard_gate * 8  # Simplified cost model
            }
        
        def _calculate_qualitative_factors(self, assignments: Dict[str, Tuple[float, float]]) -> Dict[str, float]:
            """Calculate qualitative factors"""
            # Safety: Check separation distances
            min_separation = 100  # Minimum 100m separation
            safety_score = 1.0
            
            facilities = list(assignments.values())
            for i in range(len(facilities)):
                for j in range(i + 1, len(facilities)):
                    distance = np.sqrt((facilities[i][0] - facilities[j][0])**2 + 
                                     (facilities[i][1] - facilities[j][1])**2)
                    if distance < min_separation:
                        safety_score *= (distance / min_separation)
            
            # Flexibility: Check expansion space
            terminal_bounds = (1000, 800)
            edge_buffer = 50  # 50m buffer from edges
            flexibility_score = 1.0
            
            for coord in facilities:
                if (coord[0] < edge_buffer or coord[0] > terminal_bounds[0] - edge_buffer or
                    coord[1] < edge_buffer or coord[1] > terminal_bounds[1] - edge_buffer):
                    flexibility_score *= 0.9
            
            # Environmental impact (simplified)
            environmental_score = 0.8 + 0.2 * random.random()  # Base score with variation
            
            # Regulatory compliance
            regulatory_score = 0.9 + 0.1 * random.random()  # High compliance with variation
            
            return {
                'safety_score': safety_score * 100,
                'flexibility_score': flexibility_score * 100,
                'environmental_score': environmental_score * 100,
                'regulatory_compliance': regulatory_score * 100
            }
        
        def _calculate_ai_confidence(self, quantitative: Dict[str, float], 
                                    qualitative: Dict[str, float]) -> float:
            """Calculate AI confidence in the solution"""
            # Base confidence on optimization quality
            throughput_weight = 0.4
            safety_weight = 0.3
            efficiency_weight = 0.3
            
            normalized_throughput = min(1.0, quantitative['throughput_score'] / 100)
            normalized_safety = qualitative['safety_score'] / 100
            normalized_efficiency = quantitative['transport_efficiency'] / 100
            
            confidence = (throughput_weight * normalized_throughput +
                         safety_weight * normalized_safety +
                         efficiency_weight * normalized_efficiency)
            
            return confidence
        
        def generate_explanation(self, solution: LayoutSolution) -> Dict[str, Any]:
            """Generate explainable reasoning for the solution"""
            explanation = {
                'primary_objective': 'maximize_throughput',
                'key_factors': [],
                'reasoning_steps': [],
                'tradeoffs': [],
                'confidence_breakdown': {}
            }
            
            # Identify key factors
            if solution.quantitative_metrics['throughput_score'] > 80:
                explanation['key_factors'].append('high_throughput_design')
                explanation['reasoning_steps'].append(
                    f"Optimized facility placement achieves throughput score of {solution.quantitative_metrics['throughput_score']:.1f}"
                )
            
            if solution.qualitative_factors['safety_score'] > 85:
                explanation['key_factors'].append('safety_optimization')
                explanation['reasoning_steps'].append(
                    f"Safety score of {solution.qualitative_factors['safety_score']:.1f}% through proper facility separation"
                )
            
            if solution.quantitative_metrics['transport_efficiency'] > 70:
                explanation['key_factors'].append('transport_efficiency')
                explanation['reasoning_steps'].append(
                    f"Transport efficiency of {solution.quantitative_metrics['transport_efficiency']:.1f}% minimizes internal distances"
                )
            
            # Identify tradeoffs
            if solution.quantitative_metrics['estimated_cost'] > 5000:
                explanation['tradeoffs'].append(
                    f"Higher estimated cost (${solution.quantitative_metrics['estimated_cost']:.0f}) for improved performance"
                )
            
            if solution.qualitative_factors['flexibility_score'] < 80:
                explanation['tradeoffs'].append(
                    f"Reduced flexibility ({solution.qualitative_factors['flexibility_score']:.1f}%) for optimized current operations"
                )
            
            # Confidence breakdown
            explanation['confidence_breakdown'] = {
                'throughput_contribution': solution.quantitative_metrics['throughput_score'] / 100,
                'safety_contribution': solution.qualitative_factors['safety_score'] / 100,
                'efficiency_contribution': solution.quantitative_metrics['transport_efficiency'] / 100
            }
            
            return explanation
    
    print("Explainable AI optimizer class defined successfully")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    class HumanAICollaborationSystem:
        """Human-AI collaborative decision-making system"""
        
        def __init__(self):
            self.ai_optimizer = ExplainableAIOptimizer()
            self.experts = self._initialize_experts()
            self.collaboration_history = []
            self.learning_data = []
            
        def _initialize_experts(self) -> List[HumanExpert]:
            """Initialize human experts with different expertise areas"""
            experts = [
                HumanExpert(
                    name="Dr. Schmidt",
                    expertise_areas=['safety', 'operations'],
                    experience_years=25,
                    decision_weights={'throughput': 0.3, 'safety': 0.4, 'cost': 0.2, 'flexibility': 0.1}
                ),
                HumanExpert(
                    name="Ms. Chen",
                    expertise_areas=['environmental', 'regulatory'],
                    experience_years=20,
                    decision_weights={'throughput': 0.2, 'environmental': 0.3, 'regulatory': 0.3, 'cost': 0.2}
                ),
                HumanExpert(
                    name="Mr. Rodriguez",
                    expertise_areas=['operations', 'maintenance'],
                    experience_years=30,
                    decision_weights={'throughput': 0.4, 'operational_efficiency': 0.3, 'maintenance': 0.2, 'flexibility': 0.1}
                )
            ]
            return experts
        
        def start_collaboration_session(self, expert: HumanExpert, 
                                     project_requirements: Dict[str, Any]) -> CollaborationSession:
            """Start a new human-AI collaboration session"""
            session_id = f"session_{len(self.collaboration_history) + 1}_{int(time.time())}"
            
            session = CollaborationSession(
                session_id=session_id,
                human_expert=expert,
                ai_recommendations=[],
                human_feedback=[]
            )
            
            print(f"Starting collaboration session {session_id}")
            print(f"Expert: {expert.name} (Expertise: {', '.join(expert.expertise_areas)})")
            print(f"Requirements: {project_requirements}")
            
            return session
        
        def generate_ai_recommendations(self, session: CollaborationSession, 
                                       requirements: Dict[str, Any]) -> List[AIRecommendation]:
            """Generate AI recommendations with explanations"""
            print("\nGenerating AI recommendations...")
            
            # Generate multiple AI solutions
            ai_solutions = self.ai_optimizer.generate_ai_solutions(num_solutions=15)
            
            # Rank solutions based on expert preferences
            ranked_solutions = self._rank_solutions_for_expert(ai_solutions, session.human_expert)
            
            # Create recommendations for top solutions
            recommendations = []
            for i, solution in enumerate(ranked_solutions[:3]):  # Top 3 recommendations
                explanation = self.ai_optimizer.generate_explanation(solution)
                solution.explanation = explanation
                
                # Calculate feature importance
                feature_importance = self._calculate_feature_importance(solution, session.human_expert)
                
                recommendation = AIRecommendation(
                    solution=solution,
                    primary_objective=explanation['primary_objective'],
                    feature_importance=feature_importance,
                    reasoning=explanation['reasoning_steps'],
                    alternatives=ranked_solutions[3:6],  # Alternative solutions
                    confidence_score=solution.ai_confidence
                )
                
                recommendations.append(recommendation)
                
                print(f"  Recommendation {i+1}: {explanation['primary_objective']} "
                      f"(Confidence: {solution.ai_confidence:.2f})")
            
            session.ai_recommendations = recommendations
            return recommendations
        
        def _rank_solutions_for_expert(self, solutions: List[LayoutSolution], 
                                      expert: HumanExpert) -> List[LayoutSolution]:
            """Rank solutions based on expert preferences"""
            scored_solutions = []
            
            for solution in solutions:
                # Calculate expert-specific score
                score = 0.0
                
                # Throughput consideration
                if 'throughput' in expert.decision_weights:
                    score += expert.decision_weights['throughput'] * (solution.quantitative_metrics['throughput_score'] / 100)
                
                # Safety consideration
                if 'safety' in expert.decision_weights:
                    score += expert.decision_weights['safety'] * (solution.qualitative_factors['safety_score'] / 100)
                
                # Cost consideration
                if 'cost' in expert.decision_weights:
                    # Lower cost is better, so invert
                    normalized_cost = 1.0 - min(1.0, solution.quantitative_metrics['estimated_cost'] / 10000)
                    score += expert.decision_weights['cost'] * normalized_cost
                
                # Flexibility consideration
                if 'flexibility' in expert.decision_weights:
                    score += expert.decision_weights['flexibility'] * (solution.qualitative_factors['flexibility_score'] / 100)
                
                # Environmental consideration
                if 'environmental' in expert.decision_weights:
                    score += expert.decision_weights['environmental'] * (solution.qualitative_factors['environmental_score'] / 100)
                
                # Operational efficiency
                if 'operational_efficiency' in expert.decision_weights:
                    score += expert.decision_weights['operational_efficiency'] * (solution.quantitative_metrics['transport_efficiency'] / 100)
                
                scored_solutions.append((score, solution))
            
            # Sort by score (descending)
            scored_solutions.sort(key=lambda x: x[0], reverse=True)
            
            return [solution for score, solution in scored_solutions]
        
        def _calculate_feature_importance(self, solution: LayoutSolution, 
                                         expert: HumanExpert) -> Dict[str, float]:
            """Calculate feature importance for the solution"""
            importance = {}
            
            # Quantitative features
            importance['throughput_score'] = solution.quantitative_metrics['throughput_score'] / 100
            importance['transport_efficiency'] = solution.quantitative_metrics['transport_efficiency'] / 100
            importance['cost_efficiency'] = 1.0 - min(1.0, solution.quantitative_metrics['estimated_cost'] / 10000)
            
            # Qualitative features
            importance['safety'] = solution.qualitative_factors['safety_score'] / 100
            importance['flexibility'] = solution.qualitative_factors['flexibility_score'] / 100
            importance['environmental'] = solution.qualitative_factors['environmental_score'] / 100
            importance['regulatory'] = solution.qualitative_factors['regulatory_compliance'] / 100
            
            # Weight by expert preferences
            for feature in importance:
                if feature in expert.decision_weights:
                    importance[feature] *= expert.decision_weights[feature]
            
            # Normalize to sum to 1
            total_importance = sum(importance.values())
            if total_importance > 0:
                importance = {k: v/total_importance for k, v in importance.items()}
            
            return importance
        
        def collect_human_feedback(self, session: CollaborationSession, 
                                  recommendation: AIRecommendation, 
                                  feedback: Dict[str, Any]) -> None:
            """Collect and process human expert feedback"""
            print(f"\nCollecting feedback from {session.human_expert.name}...")
            
            # Record feedback
            feedback_record = {
                'timestamp': time.time(),
                'recommendation_id': id(recommendation),
                'feedback': feedback,
                'expert_concerns': feedback.get('concerns', []),
                'suggested_modifications': feedback.get('modifications', []),
                'acceptance_level': feedback.get('acceptance_level', 0.5)
            }
            
            session.human_feedback.append(feedback_record)
            
            # Update solution with human acceptance
            recommendation.solution.human_acceptance = feedback['acceptance_level']
            
            print(f"  Acceptance level: {feedback['acceptance_level']:.1f}")
            if feedback.get('concerns'):
                print(f"  Concerns: {', '.join(feedback['concerns'])}")
            if feedback.get('modifications'):
                print(f"  Suggested modifications: {', '.join(feedback['modifications'])}")
        
        def collaborative_refinement(self, session: CollaborationSession) -> LayoutSolution:
            """Perform collaborative refinement of AI solutions based on human feedback"""
            print("\nPerforming collaborative refinement...")
            
            # Get the best recommendation
            best_recommendation = session.ai_recommendations[0]
            base_solution = best_recommendation.solution
            
            # Apply human feedback modifications
            refined_solution = self._apply_human_modifications(base_solution, session)
            
            # Recalculate metrics for refined solution
            refined_solution.quantitative_metrics = self.ai_optimizer._calculate_quantitative_metrics(
                refined_solution.facility_assignments)
            refined_solution.qualitative_factors = self.ai_optimizer._calculate_qualitative_factors(
                refined_solution.facility_assignments)
            
            # Generate new explanation for refined solution
            refined_solution.explanation = self.ai_optimizer.generate_explanation(refined_solution)
            
            session.final_solution = refined_solution
            
            print(f"Refinement completed. Final solution acceptance: {refined_solution.human_acceptance:.2f}")
            
            return refined_solution
        
        def _apply_human_modifications(self, solution: LayoutSolution, 
                                    session: CollaborationSession) -> LayoutSolution:
            """Apply human expert modifications to the solution"""
            # Create a copy of the solution
            refined_assignments = solution.facility_assignments.copy()
            
            # Apply modifications based on human feedback
            for feedback in session.human_feedback:
                modifications = feedback.get('suggested_modifications', [])
                concerns = feedback.get('expert_concerns', [])
                
                # Apply safety-related modifications
                if 'safety' in concerns or 'narrow_aisles' in concerns:
                    # Increase separation between facilities
                    refined_assignments = self._increase_facility_separation(refined_assignments)
                
                # Apply regulatory compliance modifications
                if 'regulatory' in concerns or 'setback_violation' in concerns:
                    # Ensure proper setbacks from boundaries
                    refined_assignments = self._ensure_regulatory_setbacks(refined_assignments)
                
                # Apply operational modifications
                if 'operational' in concerns or 'access_efficiency' in modifications:
                    # Optimize access routes
                    refined_assignments = self._optimize_access_routes(refined_assignments)
            
            # Create refined solution
            refined_solution = LayoutSolution(
                facility_assignments=refined_assignments,
                quantitative_metrics=solution.quantitative_metrics.copy(),
                qualitative_factors=solution.qualitative_factors.copy(),
                ai_confidence=solution.ai_confidence * 0.9,  # Slightly reduced confidence after modifications
                human_acceptance=0.85  # Improved human acceptance after addressing concerns
            )
            
            return refined_solution
        
        def _increase_facility_separation(self, assignments: Dict[str, Tuple[float, float]]) -> Dict[str, Tuple[float, float]]:
            """Increase separation between facilities for safety"""
            refined = assignments.copy()
            
            # Increase minimum separation from 100m to 150m
            min_separation = 150
            
            # Simple approach: move facilities away from each other if too close
            facility_names = list(refined.keys())
            for i in range(len(facility_names)):
                for j in range(i + 1, len(facility_names)):
                    name1, name2 = facility_names[i], facility_names[j]
                    coord1, coord2 = refined[name1], refined[name2]
                    
                    distance = np.sqrt((coord1[0] - coord2[0])**2 + (coord1[1] - coord2[1])**2)
                    
                    if distance < min_separation:
                        # Move facilities apart
                        if coord1[0] < coord2[0]:  # name1 is to the left
                            refined[name1] = (max(0, coord1[0] - 20), coord1[1])
                            refined[name2] = (min(1000, coord2[0] + 20), coord2[1])
                        else:
                            refined[name1] = (min(1000, coord1[0] + 20), coord1[1])
                            refined[name2] = (max(0, coord2[0] - 20), coord2[1])
            
            return refined
        
        def _ensure_regulatory_setbacks(self, assignments: Dict[str, Tuple[float, float]]) -> Dict[str, Tuple[float, float]]:
            """Ensure regulatory setbacks from boundaries and sensitive areas"""
            refined = assignments.copy()
            
            # Regulatory setback: 50m from all boundaries
            setback = 50
            terminal_width = 1000
            terminal_length = 800
            
            for name, coord in refined.items():
                x, y = coord
                
                # Apply setbacks
                new_x = max(setback, min(terminal_width - setback, x))
                new_y = max(setback, min(terminal_length - setback, y))
                
                refined[name] = (new_x, new_y)
            
            return refined
        
        def _optimize_access_routes(self, assignments: Dict[str, Tuple[float, float]]) -> Dict[str, Tuple[float, float]]:
            """Optimize access routes for operational efficiency"""
            refined = assignments.copy()
            
            # Ensure gates have good access to main roads
            # Move gates closer to edges for truck access
            for name, coord in refined.items():
                if 'gate' in name:
                    x, y = coord
                    # Move gates closer to landside edge (y = 800)
                    if y < 600:  # If gate is not already landside
                        new_y = min(750, y + 50)  # Move towards landside
                        refined[name] = (x, new_y)
            
            return refined
        
        def evaluate_collaboration_outcome(self, session: CollaborationSession) -> Dict[str, Any]:
            """Evaluate the outcome of the human-AI collaboration"""
            if not session.final_solution:
                return {'error': 'No final solution available'}
            
            original_solution = session.ai_recommendations[0].solution
            refined_solution = session.final_solution
            
            # Calculate improvements
            throughput_change = (refined_solution.quantitative_metrics['throughput_score'] - 
                               original_solution.quantitative_metrics['throughput_score'])
            safety_change = (refined_solution.qualitative_factors['safety_score'] - 
                            original_solution.qualitative_factors['safety_score'])
            acceptance_improvement = (refined_solution.human_acceptance - 
                                     (original_solution.human_acceptance or 0.5))
            
            outcome = {
                'throughput_improvement': throughput_change,
                'safety_improvement': safety_change,
                'acceptance_improvement': acceptance_improvement,
                'final_acceptance': refined_solution.human_acceptance,
                'collaboration_success': refined_solution.human_acceptance > 0.8,
                'num_feedback_iterations': len(session.human_feedback),
                'expert_satisfaction': session.human_expert.name
            }
            
            return outcome
    
    print("Human-AI collaboration system defined successfully")
except Exception as e:
    print(f'Error in cell: {e}')
```
