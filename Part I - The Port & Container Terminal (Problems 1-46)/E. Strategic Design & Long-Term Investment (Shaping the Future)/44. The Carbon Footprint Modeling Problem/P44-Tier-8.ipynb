# 44. The Carbon Footprint Modeling Problem

## Tier 8: The Value-Aligned & Ethical Framework

### Goal
Implement a value-aligned carbon footprint management system that transcends pure optimization to incorporate ethical principles, stakeholder values, and long-term sustainability goals, ensuring that carbon reduction strategies align with broader societal values including environmental justice, economic equity, and intergenerational responsibility.

### Key assumptions
- Carbon optimization must balance multiple stakeholder interests
- Ethical constraints are as important as efficiency metrics
- Long-term sustainability outweighs short-term cost savings
- Environmental justice and equity are non-negotiable requirements
- Stakeholder participation enhances decision legitimacy

### Approach (step-by-step)
1. Define multi-layered ethical framework with constitutional principles
2. Implement stakeholder value balancing mechanisms
3. Create temporal sustainability vs short-term efficiency trade-offs
4. Develop environmental impact distribution and fairness analysis
5. Build ethical constraint enforcement and violation detection
6. Implement stakeholder feedback integration systems
7. Create value-aligned scoring and decision recommendation
8. Validate system with real-world ethical scenarios

### What to look for in the results
- Ethical constraint compliance and violation detection
- Stakeholder value balancing and trade-off analysis
- Environmental justice impact distribution
- Value-aligned scoring vs pure optimization metrics
- Long-term sustainability vs short-term efficiency analysis
- Ethical decision recommendations with transparency

### Concrete example (from the source)
Value-aligned optimization for a regional logistics network serving 3 communities:

Stakeholder Value Weights:
- Community health: 0.35
- Worker security: 0.30
- Shareholder returns: 0.15
- Environmental justice: 0.20

Scenario Evaluation Results:

Scenario A - Pure Carbon Optimization:
- Carbon reduction: 28% (-450 kg CO₂e/day)
- Community impact: +0.65 (air quality improvement)
- Worker impact: -0.40 (120 job losses)
- Ethical compliance: VIOLATION (hard constraint on job losses)
- Value-aligned score: 0.0 (rejected)

Scenario B - Balanced Transition:
- Carbon reduction: 22% (-354 kg CO₂e/day)
- Community impact: +0.55 (moderate air quality improvement)
- Worker impact: +0.25 (net 30 job gains through retraining programs)
- Ethical compliance: FULL COMPLIANCE
- Value-aligned score: 0.78 (selected)

Scenario C - Minimal Change:
- Carbon reduction: 8% (-129 kg CO₂e/day)
- Community impact: +0.20 (limited air quality improvement)
- Worker impact: +0.80 (no job losses, high security)
- Ethical compliance: FULL COMPLIANCE
- Value-aligned score: 0.52 (suboptimal)

Ethical Decision: The system selects Scenario B, achieving 354 kg CO₂e daily reduction while maintaining full ethical compliance and supporting worker transition. The solution demonstrates that value-aligned optimization can achieve substantial carbon reduction (78% of maximum possible) while respecting stakeholder values and ethical constraints.

```python
# Import required libraries for Value-Aligned & Ethical Framework
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass
from typing import List, Tuple, Dict, Optional, Any
import random
from enum import Enum
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('default')
sns.set_palette("husl")
```

```python
class EthicalPrinciple(Enum):
    """Core ethical principles for carbon footprint management"""
    ENVIRONMENTAL_JUSTICE = "environmental_justice"
    INTERGENERATIONAL_EQUITY = "intergenerational_equity"
    WORKER_PROTECTION = "worker_protection"
    COMMUNITY_HEALTH = "community_health"
    ECONOMIC_SUSTAINABILITY = "economic_sustainability"
    TRANSPARENCY = "transparency"

@dataclass
class StakeholderGroup:
    """Represents a stakeholder group with values and weights"""
    name: str
    weight: float
    values: Dict[str, float]
    impact_sensitivity: float

@dataclass
class EthicalConstraint:
    """Represents an ethical constraint with threshold"""
    principle: EthicalPrinciple
    threshold: float
    is_hard_constraint: bool
    description: str

@dataclass
class ScenarioImpact:
    """Represents the impact of a scenario on different aspects"""
    carbon_reduction: float  # kg CO₂e/day
    community_health: float
    worker_security: float
    shareholder_returns: float
    environmental_justice: float
    long_term_sustainability: float

@dataclass
class EthicalScenario:
    """Represents an ethical scenario for evaluation"""
    id: str
    name: str
    description: str
    cost_increase: float  # percentage
    implementation_time: str
    impact: ScenarioImpact
    value_aligned_score: float
    compliance_status: str
```

```python
class ValueAlignedOptimization:
    """Value-Aligned Carbon Footprint Optimization System"""
    
    def __init__(self):
        self.stakeholders = self._initialize_stakeholders()
        self.constraints = self._initialize_constraints()
        self.scenarios = []
    
    def _initialize_stakeholders(self) -> List[StakeholderGroup]:
        """Initialize stakeholder groups with their values and weights"""
        return [
            StakeholderGroup(
                name="Community",
                weight=0.35,
                values={
                    "air_quality": 0.8,
                    "noise_pollution": 0.6,
                    "green_spaces": 0.7,
                    "property_values": 0.5
                },
                impact_sensitivity=0.8
            ),
            StakeholderGroup(
                name="Workers",
                weight=0.30,
                values={
                    "job_security": 0.9,
                    "wage_levels": 0.7,
                    "working_conditions": 0.8,
                    "training_opportunities": 0.6
                },
                impact_sensitivity=0.9
            ),
            StakeholderGroup(
                name="Shareholders",
                weight=0.15,
                values={
                    "profitability": 0.8,
                    "market_share": 0.6,
                    "risk_management": 0.7,
                    "brand_reputation": 0.5
                },
                impact_sensitivity=0.6
            ),
            StakeholderGroup(
                name="Environment",
                weight=0.20,
                values={
                    "carbon_reduction": 0.9,
                    "biodiversity": 0.8,
                    "water_quality": 0.7,
                    "soil_health": 0.6
                },
                impact_sensitivity=0.7
            )
        ]
    
    def _initialize_constraints(self) -> List[EthicalConstraint]:
        """Initialize ethical constraints"""
        return [
            EthicalConstraint(
                principle=EthicalPrinciple.WORKER_PROTECTION,
                threshold=-0.2,  # Maximum 20% job loss
                is_hard_constraint=True,
                description="Worker job losses cannot exceed 20%"
            ),
            EthicalConstraint(
                principle=EthicalPrinciple.ENVIRONMENTAL_JUSTICE,
                threshold=-0.3,  # Minimum 30% positive impact on disadvantaged communities
                is_hard_constraint=True,
                description="Must provide minimum 30% benefit to disadvantaged communities"
            ),
            EthicalConstraint(
                principle=EthicalPrinciple.COMMUNITY_HEALTH,
                threshold=0.1,  # Minimum 10% improvement in community health
                is_hard_constraint=False,
                description="Should provide minimum 10% improvement in community health"
            ),
            EthicalConstraint(
                principle=EthicalPrinciple.ECONOMIC_SUSTAINABILITY,
                threshold=-0.5,  # Maximum 50% cost increase
                is_hard_constraint=False,
                description="Cost increase should not exceed 50%"
            )
        ]
    
    def create_scenarios(self) -> List[EthicalScenario]:
        """Create ethical scenarios for evaluation"""
        
        scenarios = [
            EthicalScenario(
                id="pure_optimization",
                name="Pure Carbon Optimization",
                description="Maximum carbon reduction without ethical considerations",
                cost_increase=-5.0,  # Actually reduces costs
                implementation_time="short_term",
                impact=ScenarioImpact(
                    carbon_reduction=450,
                    community_health=0.65,
                    worker_security=-0.40,
                    shareholder_returns=0.30,
                    environmental_justice=0.20,
                    long_term_sustainability=0.40
                ),
                value_aligned_score=0.0,
                compliance_status="VIOLATION"
            ),
            EthicalScenario(
                id="balanced_transition",
                name="Balanced Transition",
                description="Carbon reduction with worker transition programs",
                cost_increase=2.3,
                implementation_time="medium_term",
                impact=ScenarioImpact(
                    carbon_reduction=354,
                    community_health=0.55,
                    worker_security=0.25,
                    shareholder_returns=0.10,
                    environmental_justice=0.45,
                    long_term_sustainability=0.75
                ),
                value_aligned_score=0.0,
                compliance_status="FULL_COMPLIANCE"
            ),
            EthicalScenario(
                id="minimal_change",
                name="Minimal Change",
                description="Conservative approach with minimal disruption",
                cost_increase=0.8,
                implementation_time="short_term",
                impact=ScenarioImpact(
                    carbon_reduction=129,
                    community_health=0.20,
                    worker_security=0.80,
                    shareholder_returns=-0.10,
                    environmental_justice=0.15,
                    long_term_sustainability=0.30
                ),
                value_aligned_score=0.0,
                compliance_status="VIOLATION"
            ),
            EthicalScenario(
                id="community_first",
                name="Community-First Approach",
                description="Prioritize community benefits and environmental justice",
                cost_increase=4.5,
                implementation_time="long_term",
                impact=ScenarioImpact(
                    carbon_reduction=280,
                    community_health=0.80,
                    worker_security=0.10,
                    shareholder_returns=-0.15,
                    environmental_justice=0.85,
                    long_term_sustainability=0.70
                ),
                value_aligned_score=0.0,
                compliance_status="FULL_COMPLIANCE"
            ),
            EthicalScenario(
                id="worker_protection",
                name="Worker Protection Focus",
                description="Maximum worker protection with moderate carbon goals",
                cost_increase=3.2,
                implementation_time="short_term",
                impact=ScenarioImpact(
                    carbon_reduction=180,
                    community_health=0.35,
                    worker_security=0.90,
                    shareholder_returns=-0.10,
                    environmental_justice=0.40,
                    long_term_sustainability=0.50
                ),
                value_aligned_score=0.0,
                compliance_status="VIOLATION"
            )
        ]
        
        self.scenarios = scenarios
        return scenarios
    
    def check_ethical_compliance(self, scenario: EthicalScenario) -> Tuple[str, List[str]]:
        """Check if scenario complies with ethical constraints"""
        violations = []
        
        for constraint in self.constraints:
            if constraint.principle == EthicalPrinciple.WORKER_PROTECTION:
                if scenario.impact.worker_security < constraint.threshold:
                    violations.append(f"Worker protection: {scenario.impact.worker_security:.1f} < {constraint.threshold:.1f}")
            
            elif constraint.principle == EthicalPrinciple.ENVIRONMENTAL_JUSTICE:
                if scenario.impact.environmental_justice < constraint.threshold:
                    violations.append(f"Environmental justice: {scenario.impact.environmental_justice:.1f} < {constraint.threshold:.1f}")
            
            elif constraint.principle == EthicalPrinciple.COMMUNITY_HEALTH:
                if scenario.impact.community_health < constraint.threshold:
                    violations.append(f"Community health: {scenario.impact.community_health:.1f} < {constraint.threshold:.1f}")
            
            elif constraint.principle == EthicalPrinciple.ECONOMIC_SUSTAINABILITY:
                if scenario.cost_increase > abs(constraint.threshold) * 100:
                    violations.append(f"Economic sustainability: {scenario.cost_increase:.1f}% > {abs(constraint.threshold)*100:.1f}%")
        
        # Check for minimum carbon reduction
        if scenario.impact.carbon_reduction < 100:  # Minimum 100 kg CO₂e/day
            violations.append(f"Insufficient carbon reduction: {scenario.impact.carbon_reduction:.0f} < 100 kg CO₂e/day")
        
        compliance_status = "VIOLATION" if violations else "FULL_COMPLIANCE"
        return compliance_status, violations
    
    def calculate_value_aligned_score(self, scenario: EthicalScenario) -> float:
        """Calculate value-aligned score for a scenario"""
        
        # Check compliance first
        compliance_status, violations = self.check_ethical_compliance(scenario)
        
        # If hard constraints violated, score is 0
        hard_constraint_violations = [v for v in violations if any(c.principle.value in v for c in self.constraints if c.is_hard_constraint)]
        if hard_constraint_violations:
            return 0.0
        
        # Calculate weighted stakeholder score
        stakeholder_scores = []
        
        for stakeholder in self.stakeholders:
            if stakeholder.name == "Community":
                score = (scenario.impact.community_health * 0.4 + 
                        scenario.impact.environmental_justice * 0.4 + 
                        scenario.impact.long_term_sustainability * 0.2)
            
            elif stakeholder.name == "Workers":
                score = scenario.impact.worker_security
            
            elif stakeholder.name == "Shareholders":
                score = (scenario.impact.shareholder_returns * 0.6 + 
                        (1 - abs(scenario.cost_increase) / 100) * 0.4)
            
            elif stakeholder.name == "Environment":
                score = (scenario.impact.carbon_reduction / 450 * 0.5 +  # Normalized to max reduction
                        scenario.impact.environmental_justice * 0.3 + 
                        scenario.impact.long_term_sustainability * 0.2)
            
            stakeholder_scores.append(score * stakeholder.weight)
        
        # Apply penalty for soft constraint violations
        soft_constraint_violations = [v for v in violations if not any(c.principle.value in v for c in self.constraints if c.is_hard_constraint)]
        penalty = len(soft_constraint_violations) * 0.1
        
        # Calculate final score
        base_score = sum(stakeholder_scores)
        final_score = max(0.0, base_score - penalty)
        
        return min(1.0, final_score)  # Cap at 1.0
    
    def evaluate_scenarios(self) -> List[EthicalScenario]:
        """Evaluate all scenarios and calculate scores"""
        
        # Check compliance and update status
        for scenario in self.scenarios:
            compliance_status, violations = self.check_ethical_compliance(scenario)
            scenario.compliance_status = compliance_status
        
        return self.scenarios

# Create the ethical optimization system
ethical_optimizer = ValueAlignedOptimization()
scenarios = ethical_optimizer.create_scenarios()
scenarios = ethical_optimizer.evaluate_scenarios()

print(f"Created {len(scenarios)} ethical scenarios for evaluation")
print(f"Stakeholder groups: {[s.name for s in ethical_optimizer.stakeholders]}")
print(f"Ethical constraints: {len(ethical_optimizer.constraints)}")
```

```python
def analyze_ethical_scenarios(scenarios: List[EthicalScenario]) -> Dict:
    """Analyze and display ethical scenario evaluation results"""

    print("\n" + "=" * 80)
    print("ETHICAL SCENARIO EVALUATION RESULTS")
    print("=" * 80)

    # Sort scenarios by value-aligned score
    sorted_scenarios = sorted(scenarios, key=lambda x: x.value_aligned_score, reverse=True)

    print(f"\n--- SCENARIO RANKING BY VALUE-ALIGNED SCORE ---")
    print(f"{'Rank':<6} {'Scenario':<20} {'Score':<8} {'Compliance':<15} {'Carbon Reduction':<18}")
    print("-" * 75)

    for i, scenario in enumerate(sorted_scenarios):
        print(f"{i+1:<6} {scenario.name[:18]:<20} {scenario.value_aligned_score:<8.3f} "
              f"{scenario.compliance_status:<15} {scenario.impact.carbon_reduction:>8.0f} kg CO₂e/day")

    # Detailed analysis of top scenarios
    print(f"\n--- DETAILED ANALYSIS OF TOP SCENARIOS ---")

    for scenario in sorted_scenarios[:3]:
        print(f"\n{scenario.name}:")
        print(f"  Description: {scenario.description}")
        print(f"  Value-Aligned Score: {scenario.value_aligned_score:.3f}")
        print(f"  Ethical Compliance: {scenario.compliance_status}")
        print(f"  Cost Impact: {scenario.cost_increase:+.1f}%")
        print(f"  Implementation: {scenario.implementation_time}")

        print(f"  Stakeholder Impacts:")
        print(f"    Community Health: {scenario.impact.community_health:+.2f}")
        print(f"    Worker Security: {scenario.impact.worker_security:+.2f}")
        print(f"    Shareholder Returns: {scenario.impact.shareholder_returns:+.2f}")
        print(f"    Environmental Justice: {scenario.impact.environmental_justice:+.2f}")
        print(f"    Long-term Sustainability: {scenario.impact.long_term_sustainability:+.2f}")

        # Check compliance details
        compliance_status, violations = ethical_optimizer.check_ethical_compliance(scenario)
        if violations:
            print(f"  Ethical Violations:")
            for violation in violations:
                print(f"    - {violation}")

    # Compare with expected results (from source material)
    print(f"\n--- COMPARISON WITH EXPECTED RESULTS ---")

    expected_scenarios = {
        "pure_optimization": {"score": 0.0, "compliance": "VIOLATION", "carbon": 450},
        "balanced_transition": {"score": 0.78, "compliance": "FULL_COMPLIANCE", "carbon": 354},
        "minimal_change": {"score": 0.52, "compliance": "FULL_COMPLIANCE", "carbon": 129}
    }

    for scenario_id, expected in expected_scenarios.items():
        scenario = next((s for s in scenarios if s.id == scenario_id), None)
        if scenario:
            score_accuracy = (1 - abs(scenario.value_aligned_score - expected['score']) / expected['score']) * 100 if expected['score'] > 0 else 100
            carbon_accuracy = (1 - abs(scenario.impact.carbon_reduction - expected['carbon']) / expected['carbon']) * 100
            
            print(f"{scenario.name} - Expected vs Actual:")
            print(f"  Score: Expected {expected['score']:.2f}, Actual {scenario.value_aligned_score:.3f}, Accuracy {score_accuracy:.1f}%")
            print(f"  Carbon: Expected {expected['carbon']} kg, Actual {scenario.impact.carbon_reduction:.0f} kg, Accuracy {carbon_accuracy:.1f}%")
            print(f"  Compliance: Expected {expected['compliance']}, Actual {scenario.compliance_status}")
            print()

    # Select best scenario
    best_scenario = sorted_scenarios[0]

    print(f"--- ETHICAL DECISION RECOMMENDATION ---")
    print(f"Recommended Scenario: {best_scenario.name}")
    print(f"Rationale:")
    print(f"  - Highest value-aligned score: {best_scenario.value_aligned_score:.3f}")
    print(f"  - Full ethical compliance: {best_scenario.compliance_status}")
    print(f"  - Substantial carbon reduction: {best_scenario.impact.carbon_reduction:.0f} kg CO₂e/day")
    print(f"  - Acceptable cost impact: {best_scenario.cost_increase:+.1f}%")
    print(f"  - Balanced stakeholder impacts across all groups")

    return {
        'best_scenario': best_scenario,
        'all_scenarios': sorted_scenarios,
        'ethical_compliance': {s.id: s.compliance_status for s in scenarios}
    }

# Analyze ethical scenarios
analysis_results = analyze_ethical_scenarios(scenarios)
```

```python
def visualize_ethical_analysis(scenarios: List[EthicalScenario], analysis_results: Dict):
    """Create comprehensive visualizations of ethical analysis"""
    
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('Value-Aligned & Ethical Framework Analysis', fontsize=16, fontweight='bold')
    
    # 1. Value-aligned scores comparison
    ax1 = axes[0, 0]
    
    scenario_names = [s.name[:15] for s in scenarios]
    scores = [s.value_aligned_score for s in scenarios]
    colors = ['lightgreen' if s.compliance_status == 'FULL_COMPLIANCE' else 'lightcoral' for s in scenarios]
    
    bars = ax1.bar(scenario_names, scores, color=colors, alpha=0.8, edgecolor='black')
    ax1.set_xlabel('Scenarios')
    ax1.set_ylabel('Value-Aligned Score')
    ax1.set_title('Value-Aligned Scores by Scenario')
    ax1.set_ylim(0, 1)
    ax1.grid(True, alpha=0.3, axis='y')
    
    # Add value labels on bars
    for bar, score in zip(bars, scores):
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2, height + 0.01,
                f'{score:.3f}', ha='center', va='bottom', fontweight='bold')
    
    # Add compliance indicator
    for i, scenario in enumerate(scenarios):
        if scenario.compliance_status == 'VIOLATION':
            ax1.text(i, 0.05, '✗', ha='center', va='center', fontsize=20, color='red', fontweight='bold')
        else:
            ax1.text(i, 0.05, '✓', ha='center', va='center', fontsize=20, color='green', fontweight='bold')
    
    # Rotate x-axis labels
    plt.setp(ax1.get_xticklabels(), rotation=45, ha='right')
    
    # 2. Carbon reduction vs stakeholder impacts
    ax2 = axes[0, 1]
    
    carbon_reductions = [s.impact.carbon_reduction for s in scenarios]
    worker_impacts = [s.impact.worker_security for s in scenarios]
    community_impacts = [s.impact.community_health for s in scenarios]
    
    scatter = ax2.scatter(carbon_reductions, worker_impacts, 
                       c=community_impacts, s=100, alpha=0.7, cmap='RdYlGn', edgecolors='black')
    
    # Add scenario labels
    for i, scenario in enumerate(scenarios):
        ax2.annotate(scenario.name[:10], (carbon_reductions[i], worker_impacts[i]), 
                    xytext=(5, 5), textcoords='offset points', fontsize=8)
    
    ax2.set_xlabel('Carbon Reduction (kg CO₂e/day)')
    ax2.set_ylabel('Worker Security Impact')
    ax2.set_title('Carbon Reduction vs Stakeholder Impacts')
    ax2.grid(True, alpha=0.3)
    
    # Add colorbar for community impact
    cbar = plt.colorbar(scatter, ax=ax2)
    cbar.set_label('Community Health Impact')
    
    # 3. Stakeholder impact radar chart
    ax3 = axes[1, 0]
    
    # Select top 3 scenarios for radar chart
    top_scenarios = sorted(scenarios, key=lambda x: x.value_aligned_score, reverse=True)[:3]
    
    categories = ['Community\nHealth', 'Worker\nSecurity', 'Shareholder\nReturns', 
                  'Environmental\nJustice', 'Long-term\nSustainability']
    
    # Create radar chart
    angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
    angles += angles[:1]  # Complete the circle
    
    ax3 = plt.subplot(2, 2, 3, projection='polar')
    
    colors = ['blue', 'green', 'orange']
    for i, scenario in enumerate(top_scenarios):
        values = [
            scenario.impact.community_health,
            scenario.impact.worker_security,
            scenario.impact.shareholder_returns,
            scenario.impact.environmental_justice,
            scenario.impact.long_term_sustainability
        ]
        values += values[:1]  # Complete the circle
        
        ax3.plot(angles, values, 'o-', linewidth=2, label=scenario.name[:12], color=colors[i])
        ax3.fill(angles, values, alpha=0.25, color=colors[i])
    
    ax3.set_xticks(angles[:-1])
    ax3.set_xticklabels(categories)
    ax3.set_ylim(-1, 1)
    ax3.set_title('Stakeholder Impact Comparison\n(Top 3 Scenarios)', pad=20)
    ax3.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
    ax3.grid(True)
    
    # 4. Cost vs Carbon reduction with compliance status
    ax4 = axes[1, 1]
    
    costs = [s.cost_increase for s in scenarios]
    carbon_reductions = [s.impact.carbon_reduction for s in scenarios]
    compliance_status = [s.compliance_status for s in scenarios]
    
    # Create scatter plot with different markers for compliance
    compliant_scenarios = [(c, r) for c, r, comp in zip(costs, carbon_reductions, compliance_status) if comp == 'FULL_COMPLIANCE']
    non_compliant_scenarios = [(c, r) for c, r, comp in zip(costs, carbon_reductions, compliance_status) if comp == 'VIOLATION']
    
    if compliant_scenarios:
        compliant_costs, compliant_carbon = zip(*compliant_scenarios)
        ax4.scatter(compliant_costs, compliant_carbon, s=100, c='green', alpha=0.7, 
                   edgecolors='black', linewidth=2, label='Full Compliance', marker='o')
    
    if non_compliant_scenarios:
        non_compliant_costs, non_compliant_carbon = zip(*non_compliant_scenarios)
        ax4.scatter(non_compliant_costs, non_compliant_carbon, s=100, c='red', alpha=0.7, 
                   edgecolors='black', linewidth=2, label='Violation', marker='X')
    
    # Add scenario labels
    for i, scenario in enumerate(scenarios):
        ax4.annotate(scenario.name[:10], (costs[i], carbon_reductions[i]), 
                    xytext=(5, 5), textcoords='offset points', fontsize=8)
    
    ax4.set_xlabel('Cost Increase (%)')
    ax4.set_ylabel('Carbon Reduction (kg CO₂e/day)')
    ax4.set_title('Cost vs Carbon Reduction\n(by Compliance Status)')
    ax4.grid(True, alpha=0.3)
    ax4.legend()
    
    # Add reference lines
    ax4.axhline(y=354, color='blue', linestyle='--', alpha=0.5, label='Expected Optimal')
    ax4.axvline(x=2.3, color='blue', linestyle='--', alpha=0.5)
    
    plt.tight_layout()
    plt.show()

# Visualize ethical analysis
visualize_ethical_analysis(scenarios, analysis_results)
```
