# 37. The Equipment Selection Problem (RTG vs. RMG vs. Straddle)
## Tier 8: The Value-Aligned & Ethical Framework

### Goal
Develop a comprehensive ethical decision-making framework that integrates stakeholder values, environmental justice, and social responsibility into equipment selection optimization, ensuring value-aligned outcomes that balance efficiency with ethical obligations.

### Key Assumptions
- Equipment selection decisions have significant ethical implications
- Multiple stakeholder groups have competing values and interests
- Ethical constraints can be formally encoded in optimization models
- Value-aligned decisions enhance long-term sustainability and social license

### Approach (Step-by-Step)
1. **Stakeholder Analysis**: Identify and model all affected stakeholder groups
2. **Value Integration**: Encode ethical principles as optimization constraints
3. **Multi-Criteria Decision Analysis**: Balance efficiency with ethical objectives
4. **Constitutional AI**: Implement hard constraints for unacceptable outcomes
5. **Impact Assessment**: Evaluate social and environmental consequences
6. **Stakeholder Engagement**: Incorporate participatory decision processes
7. **Ethical Optimization**: Find solutions that maximize value alignment

### What to Look for in the Results
- Stakeholder utility trade-offs and balancing mechanisms
- Ethical constraint satisfaction and violation analysis
- Social impact assessment across different dimensions
- Value alignment scores and recommendation confidence

### Concrete Example
Mediterranean Container Terminal ethical analysis:
- 280 current equipment operators affected by automation decisions
- 15,000 residents in surrounding community impacted by emissions
- RMG automation reduces employment by 45% but decreases emissions by 75%
- RTG configuration maintains employment but increases particulate emissions
- Ethical framework recommends hybrid solution with retraining programs

### Why this Tier Exists vs. Previous Tiers

**Value-Aligned Ethical Framework provides:**
- **Stakeholder Integration**: Systematic consideration of all affected parties
- **Ethical Constraints**: Hard constraints preventing unacceptable outcomes
- **Value Balance**: Quantitative trade-off analysis between competing values
- **Social License**: Enhanced legitimacy through stakeholder alignment

**Advantages over Pure Optimization (Tier 1):**
- Incorporates ethical considerations beyond cost and efficiency
- Addresses distributional justice and equity concerns
- Provides legitimacy and social acceptance for decisions
- Balances short-term efficiency with long-term sustainability

**Advantages over Technical Approaches (Tiers 2-5):**
- Explicit consideration of social and environmental impacts
- Stakeholder participation in decision processes
- Ethical constraint enforcement vs. technical optimization only
- Value alignment vs. pure performance maximization

**Advantages over AI/ML Methods (Tier 4):**
- Explicit ethical reasoning vs. black-box predictions
- Stakeholder utility modeling vs. pattern recognition
- Constitutional constraints vs. data-driven optimization
- Value alignment vs. accuracy maximization

**Advantages over Human-AI Partnership (Tier 7):**
- Systematic ethical analysis vs. intuitive human judgment
- Multi-stakeholder utility optimization vs. individual expertise
- Formal ethical constraints vs. collaborative decision making
- Quantitative value trade-offs vs. qualitative partnership

**When to Use This Tier:**
- High-impact decisions with significant social consequences
- When stakeholder conflicts require systematic resolution
- For projects requiring social license and community acceptance
- When ethical considerations are as important as efficiency

**Limitations:**
- Complex stakeholder analysis and value modeling requirements
- Subjective elements in ethical weight assignments
- May result in suboptimal pure efficiency outcomes
- Requires ongoing stakeholder engagement and validation

### Ethical Framework Data Structures

```python
@dataclass
class EquipmentType:
    name: str
    capital_cost: float
    operating_cost: float
    productivity: float
    flexibility: float
    maintenance: float
    footprint: float
    energy_consumption: float
    employment_impact: float  # Number of workers per unit
    emission_factor: float    # Tons CO2 per year per unit
    noise_level: float       # Decibels at 100m

@dataclass
class StakeholderGroup:
    name: str
    population_size: int
    primary_concerns: List[str]
    utility_weights: Dict[str, float]  # Weights for different outcomes
    veto_power: bool  # Can veto unacceptable outcomes

@dataclass
class EthicalConstraint:
    name: str
    constraint_type: str  # 'hard', 'soft', 'constitutional'
    description: str
    check_function: callable
    penalty_weight: float = 1.0

@dataclass
class ValueAlignedSolution:
    equipment_config: Dict[str, int]
    stakeholder_utilities: Dict[str, float]
    ethical_compliance: Dict[str, bool]
    overall_alignment: float
    efficiency_score: float
    recommendation_confidence: float

# Define equipment types with ethical attributes
equipment_types = {
    'RTG': EquipmentType(
        name='Rubber-Tired Gantry',
        capital_cost=2.5,
        operating_cost=0.8,
        productivity=25,
        flexibility=0.8,
        maintenance=0.3,
        footprint=200,
        energy_consumption=3.2,
        employment_impact=12.5,  # Workers per unit
        emission_factor=85.0,   # Tons CO2/year
        noise_level=75          # Decibels
    ),
    'RMG': EquipmentType(
        name='Rail-Mounted Gantry',
        capital_cost=3.2,
        operating_cost=0.6,
        productivity=35,
        flexibility=0.4,
        maintenance=0.25,
        footprint=150,
        energy_consumption=2.8,
        employment_impact=6.0,   # Workers per unit (automation reduces labor)
        emission_factor=25.0,   # Lower emissions (electric)
        noise_level=65          # Quieter operation
    ),
    'Straddle': EquipmentType(
        name='Straddle Carrier',
        capital_cost=4.1,
        operating_cost=1.1,
        productivity=20,
        flexibility=0.9,
        maintenance=0.4,
        footprint=100,
        energy_consumption=4.5,
        employment_impact=8.0,   # Workers per unit
        emission_factor=110.0,  # Higher emissions (diesel)
        noise_level=80          # Louder operation
    )
}

print('Equipment types with ethical attributes defined')
```

### Stakeholder Analysis and Value Modeling

```python
class StakeholderAnalyzer:
    def __init__(self, equipment_types: Dict[str, EquipmentType]):
        self.equipment_types = equipment_types
        self.stakeholders = {}
        self.ethical_constraints = []
        
        self._define_stakeholders()
        self._define_ethical_constraints()
    
    def _define_stakeholders(self):
        """Define stakeholder groups with their values and concerns"""
        self.stakeholders = {
            'workers': StakeholderGroup(
                name='Terminal Workers',
                population_size=280,
                primary_concerns=['employment_security', 'safety', 'skill_development'],
                utility_weights={
                    'employment_level': 0.4,
                    'workplace_safety': 0.3,
                    'training_opportunities': 0.2,
                    'job_satisfaction': 0.1
                },
                veto_power=True
            ),
            'community': StakeholderGroup(
                name='Local Community',
                population_size=15000,
                primary_concerns=['air_quality', 'noise_pollution', 'economic_development'],
                utility_weights={
                    'air_quality': 0.35,
                    'noise_level': 0.25,
                    'local_employment': 0.25,
                    'economic_growth': 0.15
                },
                veto_power=True
            ),
            'environment': StakeholderGroup(
                name='Environmental Groups',
                population_size=5000,  # Approximate membership/supporters
                primary_concerns=['carbon_emissions', 'energy_efficiency', 'sustainability'],
                utility_weights={
                    'carbon_footprint': 0.4,
                    'energy_efficiency': 0.3,
                    'sustainability_score': 0.2,
                    'renewable_energy': 0.1
                },
                veto_power=False
            ),
            'investors': StakeholderGroup(
                name='Investors/Shareholders',
                population_size=1000,  # Approximate number of investors
                primary_concerns=['profitability', 'risk_management', 'growth_potential'],
                utility_weights={
                    'financial_return': 0.5,
                    'operational_efficiency': 0.3,
                    'market_position': 0.15,
                    'risk_mitigation': 0.05
                },
                veto_power=False
            ),
            'customers': StakeholderGroup(
                name='Shipping Lines/Clients',
                population_size=50,  # Major shipping lines
                primary_concerns=['service_quality', 'cost', 'reliability'],
                utility_weights={
                    'service_cost': 0.4,
                    'service_reliability': 0.3,
                    'handling_speed': 0.2,
                    'flexibility': 0.1
                },
                veto_power=False
            )
        }
    
    def _define_ethical_constraints(self):
        """Define constitutional ethical constraints"""
        def employment_constraint(config: Dict[str, int]) -> bool:
            """No more than 40% reduction in employment"""
            current_employment = 280  # Current workers
            new_employment = sum(config[eq] * self.equipment_types[eq].employment_impact 
                               for eq in config)
            reduction_rate = (current_employment - new_employment) / current_employment
            return reduction_rate <= 0.40
        
        def emission_constraint(config: Dict[str, int]) -> bool:
            """Must achieve at least 50% emission reduction vs baseline"""
            baseline_emissions = 28000  # Tons CO2/year current
            new_emissions = sum(config[eq] * self.equipment_types[eq].emission_factor 
                              for eq in config)
            reduction_rate = (baseline_emissions - new_emissions) / baseline_emissions
            return reduction_rate >= 0.50
        
        def noise_constraint(config: Dict[str, int]) -> bool:
            """Average noise level must not exceed 70 decibels"""
            total_units = sum(config.values())
            if total_units == 0:
                return True
            
            avg_noise = sum(config[eq] * self.equipment_types[eq].noise_level 
                           for eq in config) / total_units
            return avg_noise <= 70
        
        def cost_constraint(config: Dict[str, int]) -> bool:
            """Total investment must not exceed budget"""
            budget_limit = 180.0  # Million USD
            total_cost = sum(config[eq] * self.equipment_types[eq].capital_cost 
                           for eq in config)
            return total_cost <= budget_limit
        
        self.ethical_constraints = [
            EthicalConstraint(
                name='Employment Protection',
                constraint_type='constitutional',
                description='Maximum 40% employment reduction',
                check_function=employment_constraint,
                penalty_weight=10.0
            ),
            EthicalConstraint(
                name='Emission Reduction',
                constraint_type='constitutional',
                description='Minimum 50% emission reduction required',
                check_function=emission_constraint,
                penalty_weight=8.0
            ),
            EthicalConstraint(
                name='Noise Pollution',
                constraint_type='hard',
                description='Average noise must not exceed 70dB',
                check_function=noise_constraint,
                penalty_weight=5.0
            ),
            EthicalConstraint(
                name='Budget Limit',
                constraint_type='hard',
                description='Investment must not exceed budget',
                check_function=cost_constraint,
                penalty_weight=3.0
            )
        ]
    
    def calculate_stakeholder_utility(self, config: Dict[str, int]) -> Dict[str, float]:
        """Calculate utility for each stakeholder group"""
        utilities = {}
        
        # Calculate outcome metrics
        total_employment = sum(config[eq] * self.equipment_types[eq].employment_impact 
                              for eq in config)
        total_emissions = sum(config[eq] * self.equipment_types[eq].emission_factor 
                            for eq in config)
        total_units = sum(config.values())
        avg_noise = sum(config[eq] * self.equipment_types[eq].noise_level 
                       for eq in config) / total_units if total_units > 0 else 0
        total_cost = sum(config[eq] * self.equipment_types[eq].capital_cost 
                       for eq in config)
        total_productivity = sum(config[eq] * self.equipment_types[eq].productivity 
                               for eq in config)
        
        # Calculate utilities for each stakeholder group
        for stakeholder_id, stakeholder in self.stakeholders.items():
            utility = 0.0
            weights = stakeholder.utility_weights
            
            if stakeholder_id == 'workers':
                # Workers utility based on employment, safety, training
                employment_score = min(1.0, total_employment / 280)  # Normalized to current employment
                safety_score = 0.8 + 0.2 * (1 - avg_noise / 100)  # Lower noise = higher safety
                training_score = 0.7  # Assume moderate training opportunities
                satisfaction_score = 0.6 + 0.4 * employment_score
                
                utility = (weights['employment_level'] * employment_score +
                          weights['workplace_safety'] * safety_score +
                          weights['training_opportunities'] * training_score +
                          weights['job_satisfaction'] * satisfaction_score)
            
            elif stakeholder_id == 'community':
                # Community utility based on environment and economic benefits
                air_quality_score = max(0, 1 - total_emissions / 50000)  # Normalized
                noise_score = max(0, 1 - avg_noise / 100)
                local_employment_score = min(1.0, total_employment / 300)
                economic_score = min(1.0, total_productivity / 2000)
                
                utility = (weights['air_quality'] * air_quality_score +
                          weights['noise_level'] * noise_score +
                          weights['local_employment'] * local_employment_score +
                          weights['economic_growth'] * economic_score)
            
            elif stakeholder_id == 'environment':
                # Environmental group utility
                carbon_score = max(0, 1 - total_emissions / 30000)
                energy_score = 0.7  # Assume moderate energy efficiency
                sustainability_score = carbon_score * 0.8 + 0.2  # Weight toward carbon
                renewable_score = 0.3  # Low renewable energy use
                
                utility = (weights['carbon_footprint'] * carbon_score +
                          weights['energy_efficiency'] * energy_score +
                          weights['sustainability_score'] * sustainability_score +
                          weights['renewable_energy'] * renewable_score)
            
            elif stakeholder_id == 'investors':
                # Investor utility based on financial metrics
                return_score = max(0, 1 - total_cost / 200)  # Lower cost = higher return
                efficiency_score = min(1.0, total_productivity / 2000)
                market_score = 0.7  # Moderate market position
                risk_score = 0.8  # Moderate risk mitigation
                
                utility = (weights['financial_return'] * return_score +
                          weights['operational_efficiency'] * efficiency_score +
                          weights['market_position'] * market_score +
                          weights['risk_mitigation'] * risk_score)
            
            elif stakeholder_id == 'customers':
                # Customer utility based on service quality
                cost_score = max(0, 1 - total_cost / 250)  # Lower investment = lower service cost
                reliability_score = 0.8  # Assume good reliability
                speed_score = min(1.0, total_productivity / 2000)
                flexibility_score = 0.7  # Moderate flexibility
                
                utility = (weights['service_cost'] * cost_score +
                          weights['service_reliability'] * reliability_score +
                          weights['handling_speed'] * speed_score +
                          weights['flexibility'] * flexibility_score)
            
            utilities[stakeholder_id] = utility
        
        return utilities
    
    def check_ethical_compliance(self, config: Dict[str, int]) -> Dict[str, bool]:
        """Check compliance with ethical constraints"""
        compliance = {}
        
        for constraint in self.ethical_constraints:
            compliance[constraint.name] = constraint.check_function(config)
        
        return compliance
    
    def calculate_value_alignment(self, config: Dict[str, int]) -> float:
        """Calculate overall value alignment score"""
        utilities = self.calculate_stakeholder_utility(config)
        compliance = self.check_ethical_compliance(config)
        
        # Weight stakeholders by population size and veto power
        total_weight = 0
        weighted_utility = 0
        
        for stakeholder_id, stakeholder in self.stakeholders.items():
            weight = np.log(stakeholder.population_size + 1)  # Log scale for population
            if stakeholder.veto_power:
                weight *= 2.0  # Double weight for veto power
            
            total_weight += weight
            weighted_utility += weight * utilities[stakeholder_id]
        
        # Calculate constraint compliance penalty
        penalty = 0
        for constraint in self.ethical_constraints:
            if not compliance[constraint.name]:
                penalty += constraint.penalty_weight
        
        # Normalize and apply penalty
        base_alignment = weighted_utility / total_weight if total_weight > 0 else 0
        penalty_factor = max(0, 1 - penalty / 20)  # Normalize penalty
        
        return base_alignment * penalty_factor

print('Stakeholder Analysis and Ethical Framework defined')
```

### Value-Aligned Optimization Engine

```python
class ValueAlignedOptimizer:
    def __init__(self, equipment_types: Dict[str, EquipmentType], 
                 stakeholder_analyzer: StakeholderAnalyzer):
        self.equipment_types = equipment_types
        self.stakeholder_analyzer = stakeholder_analyzer
        self.solutions = []
        self.best_solution = None
    
    def generate_configurations(self, max_units_per_type: int = 20) -> List[Dict[str, int]]:
        """Generate feasible equipment configurations"""
        configurations = []
        
        # Generate combinations with reasonable bounds
        for rtg in range(0, min(max_units_per_type, 15)):
            for rmg in range(0, min(max_units_per_type, 12)):
                for straddle in range(0, min(max_units_per_type, 18)):
                    config = {'RTG': rtg, 'RMG': rmg, 'Straddle': straddle}
                    
                    # Quick feasibility check
                    total_cost = sum(config[eq] * self.equipment_types[eq].capital_cost 
                                   for eq in config)
                    if total_cost <= 200.0:  # Reasonable cost limit
                        configurations.append(config)
        
        return configurations
    
    def evaluate_configuration(self, config: Dict[str, int]) -> ValueAlignedSolution:
        """Evaluate a configuration using value-aligned criteria"""
        # Calculate stakeholder utilities
        utilities = self.stakeholder_analyzer.calculate_stakeholder_utility(config)
        
        # Check ethical compliance
        compliance = self.stakeholder_analyzer.check_ethical_compliance(config)
        
        # Calculate overall value alignment
        alignment = self.stakeholder_analyzer.calculate_value_alignment(config)
        
        # Calculate efficiency score
        total_productivity = sum(config[eq] * self.equipment_types[eq].productivity 
                               for eq in config)
        total_cost = sum(config[eq] * self.equipment_types[eq].capital_cost 
                       for eq in config)
        efficiency = total_productivity / total_cost if total_cost > 0 else 0
        
        # Calculate recommendation confidence
        constitutional_violations = sum(1 for constraint in self.stakeholder_analyzer.ethical_constraints
                                    if constraint.constraint_type == 'constitutional' 
                                    and not compliance[constraint.name])
        
        if constitutional_violations > 0:
            confidence = 0.0  # Not acceptable
        else:
            hard_violations = sum(1 for constraint in self.stakeholder_analyzer.ethical_constraints
                               if constraint.constraint_type == 'hard' 
                               and not compliance[constraint.name])
            confidence = max(0.3, 1.0 - hard_violations * 0.2)
        
        return ValueAlignedSolution(
            equipment_config=config,
            stakeholder_utilities=utilities,
            ethical_compliance=compliance,
            overall_alignment=alignment,
            efficiency_score=efficiency,
            recommendation_confidence=confidence
    )
    
    def find_optimal_solution(self) -> ValueAlignedSolution:
        """Find the optimal value-aligned solution"""
        print('Searching for value-aligned optimal solution...')
        
        # Generate configurations
        configurations = self.generate_configurations()
        print(f'Generated {len(configurations)} configurations')
        
        # Evaluate all configurations
        solutions = []
        acceptable_solutions = []
        
        for config in configurations:
            solution = self.evaluate_configuration(config)
            solutions.append(solution)
            
            # Keep only acceptable solutions (constitutional compliance)
            if solution.recommendation_confidence > 0:
                acceptable_solutions.append(solution)
        
        print(f'Acceptable solutions: {len(acceptable_solutions)}/{len(configurations)}')
        
        # Find best solution among acceptable ones
        if acceptable_solutions:
            best_solution = max(acceptable_solutions, 
                               key=lambda s: s.overall_alignment)
        else:
            # If no acceptable solutions, find the least bad one
            best_solution = max(solutions, key=lambda s: s.overall_alignment)
        
        self.solutions = solutions
        self.best_solution = best_solution
        
        return best_solution
    
    def analyze_tradeoffs(self) -> Dict:
        """Analyze trade-offs between different objectives"""
        if not self.solutions:
            return {}
        
        # Filter acceptable solutions
        acceptable = [s for s in self.solutions if s.recommendation_confidence > 0]
        
        if not acceptable:
            return {'error': 'No acceptable solutions found'}
        
        # Find extreme solutions
        best_alignment = max(acceptable, key=lambda s: s.overall_alignment)
        best_efficiency = max(acceptable, key=lambda s: s.efficiency_score)
        most_balanced = min(acceptable, 
                          key=lambda s: max(s.stakeholder_utilities.values()) - 
                                      min(s.stakeholder_utilities.values()))
        
        return {
            'best_alignment': best_alignment,
            'best_efficiency': best_efficiency,
            'most_balanced': most_balanced,
            'total_acceptable': len(acceptable)
        }

print('Value-Aligned Optimization Engine defined')
```

### Run Value-Aligned Optimization

```python
# Create stakeholder analyzer and optimizer
stakeholder_analyzer = StakeholderAnalyzer(equipment_types)
optimizer = ValueAlignedOptimizer(equipment_types, stakeholder_analyzer)

print('=== VALUE-ALIGNED ETHICAL OPTIMIZATION ===\n')
print('Stakeholder Groups:')
for stakeholder_id, stakeholder in stakeholder_analyzer.stakeholders.items():
    print(f'  {stakeholder.name}: {stakeholder.population_size:,} people')
    print(f'    Concerns: {", ".join(stakeholder.primary_concerns)}')
    print(f'    Veto power: {stakeholder.veto_power}')

print('\nEthical Constraints:')
for constraint in stakeholder_analyzer.ethical_constraints:
    print(f'  {constraint.name} ({constraint.constraint_type}): {constraint.description}')
    print(f'    Penalty weight: {constraint.penalty_weight}')

# Find optimal solution
print('\n=== OPTIMIZATION RESULTS ===\n')
optimal_solution = optimizer.find_optimal_solution()

print(f'\nOptimal Equipment Configuration:')
for eq_type, units in optimal_solution.equipment_config.items():
    if units > 0:
        print(f'  {eq_type}: {units} units')

print(f'\nPerformance Metrics:')
print(f'  Overall value alignment: {optimal_solution.overall_alignment:.3f}')
print(f'  Efficiency score: {optimal_solution.efficiency_score:.2f}')
print(f'  Recommendation confidence: {optimal_solution.recommendation_confidence:.1%}')

print('\nStakeholder Utilities:')
for stakeholder_id, utility in optimal_solution.stakeholder_utilities.items():
    stakeholder = stakeholder_analyzer.stakeholders[stakeholder_id]
    print(f'  {stakeholder.name}: {utility:.3f}')

print('\nEthical Compliance:')
for constraint_name, compliant in optimal_solution.ethical_compliance.items():
    status = '✓ COMPLIANT' if compliant else '✗ VIOLATION'
    print(f'  {constraint_name}: {status}')

# Analyze trade-offs
tradeoffs = optimizer.analyze_tradeoffs()
if 'error' not in tradeoffs:
    print(f'\nTrade-off Analysis:')
    print(f'  Acceptable solutions: {tradeoffs["total_acceptable"]}')
    print(f'  Best alignment score: {tradeoffs["best_alignment"].overall_alignment:.3f}')
    print(f'  Best efficiency score: {tradeoffs["best_efficiency"].efficiency_score:.2f}')
```

### Impact Assessment and Analysis

```python
class ImpactAssessment:
    def __init__(self, equipment_types: Dict[str, EquipmentType], 
                 stakeholder_analyzer: StakeholderAnalyzer):
        self.equipment_types = equipment_types
        self.stakeholder_analyzer = stakeholder_analyzer
    
    def assess_employment_impact(self, config: Dict[str, int]) -> Dict:
        """Assess employment impact"""
        current_employment = 280
        new_employment = sum(config[eq] * self.equipment_types[eq].employment_impact 
                            for eq in config)
        
        change = new_employment - current_employment
        change_rate = change / current_employment
        
        return {
            'current_employment': current_employment,
            'new_employment': int(new_employment),
            'employment_change': int(change),
            'change_rate': change_rate,
            'impact_level': 'High' if abs(change_rate) > 0.3 else 'Medium' if abs(change_rate) > 0.1 else 'Low'
        }
    
    def assess_environmental_impact(self, config: Dict[str, int]) -> Dict:
        """Assess environmental impact"""
        baseline_emissions = 28000  # Current tons CO2/year
        new_emissions = sum(config[eq] * self.equipment_types[eq].emission_factor 
                          for eq in config)
        
        reduction = baseline_emissions - new_emissions
        reduction_rate = reduction / baseline_emissions
        
        # Calculate noise impact
        total_units = sum(config.values())
        avg_noise = sum(config[eq] * self.equipment_types[eq].noise_level 
                       for eq in config) / total_units if total_units > 0 else 0
        
        return {
            'baseline_emissions': baseline_emissions,
            'new_emissions': int(new_emissions),
            'emission_reduction': int(reduction),
            'reduction_rate': reduction_rate,
            'average_noise': avg_noise,
            'environmental_grade': 'A' if reduction_rate > 0.6 else 'B' if reduction_rate > 0.4 else 'C' if reduction_rate > 0.2 else 'D'
        }
    
    def assess_economic_impact(self, config: Dict[str, int]) -> Dict:
        """Assess economic impact"""
        total_investment = sum(config[eq] * self.equipment_types[eq].capital_cost 
                              for eq in config)
        total_productivity = sum(config[eq] * self.equipment_types[eq].productivity 
                               for eq in config)
        annual_operating_cost = sum(config[eq] * self.equipment_types[eq].operating_cost 
                                   for eq in config)
        
        # Estimate economic benefits
        annual_revenue_increase = total_productivity * 50  # $50 per container move
        payback_period = total_investment / annual_revenue_increase if annual_revenue_increase > 0 else float('inf')
        
        return {
            'total_investment': total_investment,
            'annual_productivity': int(total_productivity),
            'annual_operating_cost': annual_operating_cost,
            'estimated_revenue_increase': annual_revenue_increase,
            'payback_period_years': payback_period,
            'roi_percentage': (annual_revenue_increase / total_investment * 100) if total_investment > 0 else 0
        }
    
    def generate_mitigation_plan(self, config: Dict[str, int]) -> Dict:
        """Generate mitigation plan for negative impacts"""
        employment_impact = self.assess_employment_impact(config)
        environmental_impact = self.assess_environmental_impact(config)
        
        mitigation_measures = []
        total_cost = 0
        
        # Employment mitigation
        if employment_impact['employment_change'] < -50:  # Significant job losses
            mitigation_measures.append({
                'type': 'Employment',
                'measure': 'Retraining program for affected workers',
                'cost': 2.3,  # Million USD
                'effectiveness': 0.8,
                'description': 'Comprehensive retraining for new equipment operation and maintenance'
            })
            mitigation_measures.append({
                'type': 'Employment',
                'measure': 'Early retirement with incentives',
                'cost': 1.8,
                'effectiveness': 0.6,
                'description': 'Voluntary early retirement packages with skill transition support'
            })
            total_cost += 4.1
        
        # Environmental mitigation
        if environmental_impact['reduction_rate'] < 0.5:  # Insufficient emission reduction
            mitigation_measures.append({
                'type': 'Environmental',
                'measure': 'Additional emission controls',
                'cost': 3.2,
                'effectiveness': 0.7,
                'description': 'Install advanced emission control systems on diesel equipment'
            })
            mitigation_measures.append({
                'type': 'Environmental',
                'measure': 'Green energy offset program',
                'cost': 1.5,
                'effectiveness': 0.9,
                'description': 'Invest in renewable energy projects to offset remaining emissions'
            })
            total_cost += 4.7
        
        # Community mitigation
        if environmental_impact['average_noise'] > 70:
            mitigation_measures.append({
                'type': 'Community',
                'measure': 'Noise barriers and monitoring',
                'cost': 1.2,
                'effectiveness': 0.8,
                'description': 'Install noise barriers and community monitoring systems'
            })
            total_cost += 1.2
        
        return {
            'mitigation_measures': mitigation_measures,
            'total_mitigation_cost': total_cost,
            'recommended_measures': len(mitigation_measures)
        }

# Run impact assessment
impact_assessor = ImpactAssessment(equipment_types, stakeholder_analyzer)

print('=== COMPREHENSIVE IMPACT ASSESSMENT ===\n')

employment_impact = impact_assessor.assess_employment_impact(optimal_solution.equipment_config)
environmental_impact = impact_assessor.assess_environmental_impact(optimal_solution.equipment_config)
economic_impact = impact_assessor.assess_economic_impact(optimal_solution.equipment_config)
mitigation_plan = impact_assessor.generate_mitigation_plan(optimal_solution.equipment_config)

print('EMPLOYMENT IMPACT:')
print(f'  Current employment: {employment_impact["current_employment"]:,} workers')
print(f'  Projected employment: {employment_impact["new_employment"]:,} workers')
print(f'  Net change: {employment_impact["employment_change"]:+,} workers ({employment_impact["change_rate"]:+.1%})')
print(f'  Impact level: {employment_impact["impact_level"]}\n')

print('ENVIRONMENTAL IMPACT:')
print(f'  Current emissions: {employment_impact["current_employment"]:,} tons CO2/year')
print(f'  Projected emissions: {environmental_impact["new_emissions"]:,} tons CO2/year')
print(f'  Emission reduction: {environmental_impact["emission_reduction"]:+,} tons ({environmental_impact["reduction_rate"]:+.1%})')
print(f'  Average noise level: {environmental_impact["average_noise"]:.1f} dB')
print(f'  Environmental grade: {environmental_impact["environmental_grade"]}\n')

print('ECONOMIC IMPACT:')
print(f'  Total investment: ${economic_impact["total_investment"]:.1f}M')
print(f'  Annual productivity: {economic_impact["annual_productivity"]:,} moves/hour')
print(f'  Annual operating cost: ${economic_impact["annual_operating_cost"]:.1f}M')
print(f'  Estimated revenue increase: ${economic_impact["estimated_revenue_increase"]:.1f}M/year')
print(f'  Payback period: {economic_impact["payback_period_years"]:.1f} years')
print(f'  ROI: {economic_impact["roi_percentage"]:.1f}%\n')

print('MITIGATION PLAN:')
print(f'  Recommended measures: {mitigation_plan["recommended_measures"]}')
print(f'  Total mitigation cost: ${mitigation_plan["total_mitigation_cost"]:.1f}M\n')

for measure in mitigation_plan['mitigation_measures']:
    print(f'  {measure["type"]}: {measure["measure"]}')
    print(f'    Cost: ${measure["cost"]:.1f}M')
    print(f'    Effectiveness: {measure["effectiveness"]:.0%}')
    print(f'    Description: {measure["description"]}\n')
```
