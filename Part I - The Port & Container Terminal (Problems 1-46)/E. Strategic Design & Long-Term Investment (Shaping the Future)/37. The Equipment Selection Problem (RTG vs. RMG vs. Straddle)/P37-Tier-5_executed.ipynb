### Equipment and Digital Twin Data Structure

```python
@dataclass
class EquipmentType:
    name: str
    capital_cost: float  # Million USD
    operating_cost: float  # Million USD/year
    productivity: float  # Containers/hour
    flexibility: float  # 0-1 scale
    maintenance: float  # Million USD/year
    footprint: float  # Square meters
    energy_consumption: float  # kWh/container
    sensor_count: int  # Number of IoT sensors
    reliability: float  # 0-1 scale

@dataclass
class IoTSensor:
    sensor_id: str
    equipment_id: str
    sensor_type: str  # 'productivity', 'energy', 'maintenance', 'environment'
    reading_value: float
    timestamp: datetime
    accuracy: float  # 0-1 scale

@dataclass
class DigitalTwinState:
    timestamp: datetime
    equipment_states: Dict[str, Dict]  # Equipment performance data
    environmental_conditions: Dict[str, float]  # Weather, temperature, etc.
    operational_kpis: Dict[str, float]  # System performance metrics
    prediction_confidence: float  # Model confidence score

# Define equipment types with digital twin capabilities
equipment_types = {
    'RTG': EquipmentType(
        name='Rubber-Tired Gantry',
        capital_cost=2.5,
        operating_cost=0.8,
        productivity=25,
        flexibility=0.8,
        maintenance=0.3,
        footprint=200,
        energy_consumption=3.2,
        sensor_count=15,
        reliability=0.92
    ),
    'RMG': EquipmentType(
        name='Rail-Mounted Gantry',
        capital_cost=3.2,
        operating_cost=0.6,
        productivity=35,
        flexibility=0.4,
        maintenance=0.25,
        footprint=150,
        energy_consumption=2.8,
        sensor_count=20,
        reliability=0.95
    ),
    'Straddle': EquipmentType(
        name='Straddle Carrier',
        capital_cost=4.1,
        operating_cost=1.1,
        productivity=20,
        flexibility=0.9,
        maintenance=0.4,
        footprint=100,
        energy_consumption=4.5,
        sensor_count=12,
        reliability=0.88
    )
}

print('Equipment Types with Digital Twin Capabilities:')
for eq_type, eq_data in equipment_types.items():
    print(f'{eq_type}: {eq_data.name}')
    print(f'  Capital Cost: ${eq_data.capital_cost}M')
    print(f'  Productivity: {eq_data.productivity} containers/hour')
    print(f'  IoT Sensors: {eq_data.sensor_count}')
    print(f'  Reliability: {eq_data.reliability:.2f}')
    print()
```

### Digital Twin System Implementation

```python
class IntegratedDigitalTwinSystem:
    def __init__(self, equipment_types: Dict[str, EquipmentType]):
        self.equipment_types = equipment_types
        self.equipment_units = {}
        self.iot_sensors = []
        self.digital_twin_states = []
        self.prediction_models = {}
        self.scenario_results = {}
        
        # System performance metrics
        self.synchronization_accuracy = 0.95
        self.prediction_accuracy = 0.87
        self.system_uptime = 0.99
        
        self._initialize_equipment_fleet()
        self._deploy_iot_sensors()
        self._initialize_prediction_models()
        
    def _initialize_equipment_fleet(self, equipment_config: Dict[str, int] = None):
        """Initialize physical equipment fleet"""
        if equipment_config is None:
            equipment_config = {'RTG': 12, 'RMG': 8, 'Straddle': 15}
        
        unit_id_counter = 1
        for eq_type, num_units in equipment_config.items():
            for i in range(num_units):
                unit_id = f'{eq_type}_{unit_id_counter:03d}'
                
                # Initialize with realistic performance variations
                base_productivity = self.equipment_types[eq_type].productivity
                productivity_variation = np.random.normal(0, 0.1)  # 10% variation
                actual_productivity = base_productivity * (1 + productivity_variation)
                
                base_reliability = self.equipment_types[eq_type].reliability
                reliability_variation = np.random.normal(0, 0.05)
                actual_reliability = max(0.7, min(1.0, base_reliability * (1 + reliability_variation)))
                
                self.equipment_units[unit_id] = {
                    'equipment_type': eq_type,
                    'productivity': actual_productivity,
                    'reliability': actual_reliability,
                    'utilization': 0.75,
                    'maintenance_hours': 0,
                    'status': 'ACTIVE',
                    'age_years': np.random.uniform(0, 10),
                    'operating_hours': np.random.uniform(1000, 10000)
                }
                
                unit_id_counter += 1
        
        print(f'Initialized {len(self.equipment_units)} equipment units:')
        for eq_type in equipment_config:
            count = len([u for u in self.equipment_units.values() if u['equipment_type'] == eq_type])
            print(f'  {eq_type}: {count} units')
    
    def _deploy_iot_sensors(self):
        """Deploy IoT sensors across equipment fleet"""
        sensor_id_counter = 1
        
        for unit_id, unit in self.equipment_units.items():
            eq_type = unit['equipment_type']
            sensor_count = self.equipment_types[eq_type].sensor_count
            
            # Deploy different types of sensors
            sensor_types = ['productivity', 'energy', 'maintenance', 'environment', 'vibration']
            
            for i in range(sensor_count):
                sensor_type = sensor_types[i % len(sensor_types)]
                
                sensor = IoTSensor(
                    sensor_id=f'SENSOR_{sensor_id_counter:04d}',
                    equipment_id=unit_id,
                    sensor_type=sensor_type,
                    reading_value=0.0,
                    timestamp=datetime.now(),
                    accuracy=np.random.uniform(0.85, 0.98)
                )
                
                self.iot_sensors.append(sensor)
                sensor_id_counter += 1
        
        print(f'Deployed {len(self.iot_sensors)} IoT sensors across equipment fleet')
    
    def _initialize_prediction_models(self):
        """Initialize prediction models for equipment performance"""
        self.prediction_models = {
            'productivity': {
                'model_type': 'time_series',
                'accuracy': 0.87,
                'update_frequency': 'hourly'
            },
            'maintenance': {
                'model_type': 'reliability',
                'accuracy': 0.82,
                'update_frequency': 'daily'
            },
            'energy': {
                'model_type': 'consumption',
                'accuracy': 0.91,
                'update_frequency': 'real-time'
            },
            'failure': {
                'model_type': 'survival',
                'accuracy': 0.79,
                'update_frequency': 'weekly'
            }
        }
        
        print('Initialized prediction models:')
        for model_name, model_info in self.prediction_models.items():
            print(f'  {model_name}: {model_info["model_type"]} (accuracy: {model_info["accuracy"]:.2f})')
    
    def collect_sensor_data(self, current_time: datetime) -> List[IoTSensor]:
        """Collect real-time sensor data from equipment"""
        sensor_readings = []
        
        for sensor in self.iot_sensors:
            unit_id = sensor.equipment_id
            unit = self.equipment_units[unit_id]
            eq_type = unit['equipment_type']
            
            # Simulate sensor reading based on equipment state and conditions
            base_reading = self._calculate_sensor_reading(sensor.sensor_type, unit, current_time)
            
            # Add sensor noise and accuracy factors
            noise = np.random.normal(0, 0.05)  # 5% noise
            accuracy_factor = sensor.accuracy
            final_reading = base_reading * (1 + noise) * accuracy_factor
            
            # Update sensor reading
            sensor.reading_value = final_reading
            sensor.timestamp = current_time
            
            sensor_readings.append(sensor)
        
        return sensor_readings
    
    def _calculate_sensor_reading(self, sensor_type: str, unit: Dict, current_time: datetime) -> float:
        """Calculate sensor reading based on equipment state"""
        eq_type = unit['equipment_type']
        base_value = self.equipment_types[eq_type]
        
        # Time-based factors
        hour_factor = 0.7 + 0.3 * np.sin(current_time.hour * np.pi / 12)
        weather_factor = np.random.uniform(0.9, 1.1)
        
        if sensor_type == 'productivity':
            return unit['productivity'] * unit['utilization'] * hour_factor * weather_factor
        elif sensor_type == 'energy':
            return base_value.energy_consumption * unit['utilization'] * weather_factor
        elif sensor_type == 'maintenance':
            return unit['maintenance_hours'] / 100  # Normalized maintenance need
        elif sensor_type == 'environment':
            return np.random.uniform(0.8, 1.2)  # Environmental conditions
        elif sensor_type == 'vibration':
            return (1 - unit['reliability']) * 10  # Vibration indicator
        else:
            return 0.0
    
    def update_digital_twin_state(self, sensor_data: List[IoTSensor], current_time: datetime) -> DigitalTwinState:
        """Update digital twin state with latest sensor data"""
        
        # Aggregate sensor data by equipment
        equipment_states = {}
        
        for unit_id, unit in self.equipment_units.items():
            unit_sensors = [s for s in sensor_data if s.equipment_id == unit_id]
            
            # Calculate aggregated metrics
            productivity_readings = [s.reading_value for s in unit_sensors if s.sensor_type == 'productivity']
            energy_readings = [s.reading_value for s in unit_sensors if s.sensor_type == 'energy']
            maintenance_readings = [s.reading_value for s in unit_sensors if s.sensor_type == 'maintenance']
            
            equipment_states[unit_id] = {
                'current_productivity': np.mean(productivity_readings) if productivity_readings else unit['productivity'],
                'current_energy': np.mean(energy_readings) if energy_readings else 0,
                'maintenance_need': np.mean(maintenance_readings) if maintenance_readings else 0,
                'utilization': unit['utilization'],
                'reliability': unit['reliability'],
                'status': unit['status']
            }
        
        # Environmental conditions
        environmental_sensors = [s for s in sensor_data if s.sensor_type == 'environment']
        environmental_conditions = {
            'temperature': np.random.uniform(15, 35),
            'humidity': np.random.uniform(40, 80),
            'wind_speed': np.random.uniform(0, 15),
            'visibility': np.random.uniform(0.5, 1.0)
        }
        
        # Calculate system KPIs
        total_productivity = sum(state['current_productivity'] for state in equipment_states.values())
        total_energy = sum(state['current_energy'] for state in equipment_states.values())
        avg_utilization = np.mean([state['utilization'] for state in equipment_states.values()])
        active_units = len([s for s in equipment_states.values() if s['status'] == 'ACTIVE'])
        
        operational_kpis = {
            'total_productivity': total_productivity,
            'total_energy_consumption': total_energy,
            'average_utilization': avg_utilization,
            'active_equipment_count': active_units,
            'system_efficiency': total_productivity / (total_energy + 0.001),
            'maintenance_alerts': len([s for s in equipment_states.values() if s['maintenance_need'] > 0.7])
        }
        
        # Calculate prediction confidence
        prediction_confidence = np.mean([model['accuracy'] for model in self.prediction_models.values()])
        
        # Create digital twin state
        twin_state = DigitalTwinState(
            timestamp=current_time,
            equipment_states=equipment_states,
            environmental_conditions=environmental_conditions,
            operational_kpis=operational_kpis,
            prediction_confidence=prediction_confidence
        )
        
        self.digital_twin_states.append(twin_state)
        
        return twin_state
    
    def generate_predictions(self, horizon_hours: int = 24) -> Dict[str, List[float]]:
        """Generate predictions for equipment performance"""
        if not self.digital_twin_states:
            return {}
        
        latest_state = self.digital_twin_states[-1]
        predictions = {}
        
        # Generate predictions for each model type
        for model_name, model_info in self.prediction_models.items():
            model_predictions = []
            
            for hour in range(horizon_hours):
                # Time-based factors
                time_factor = 1.0 + hour * 0.01  # Gradual change
                seasonal_factor = 0.9 + 0.2 * np.sin(hour * np.pi / 12)
                
                # Base prediction on current state
                if model_name == 'productivity':
                    base_value = latest_state.operational_kpis['total_productivity']
                    prediction = base_value * time_factor * seasonal_factor
                elif model_name == 'energy':
                    base_value = latest_state.operational_kpis['total_energy_consumption']
                    prediction = base_value * time_factor * seasonal_factor
                elif model_name == 'maintenance':
                    base_value = latest_state.operational_kpis['maintenance_alerts']
                    prediction = base_value * (1 + hour * 0.02)  # Maintenance needs increase
                elif model_name == 'failure':
                    base_value = 0.05  # Base failure probability
                    prediction = base_value * (1 + hour * 0.001)  # Slight increase over time
                
                # Add model uncertainty
                uncertainty = (1 - model_info['accuracy']) * 0.1
                noise = np.random.normal(0, uncertainty)
                final_prediction = prediction * (1 + noise)
                
                model_predictions.append(max(0, final_prediction))
            
            predictions[model_name] = model_predictions
        
        return predictions
    
    def run_scenario_analysis(self, scenario_config: Dict) -> Dict[str, float]:
        """Run what-if scenario analysis"""
        scenario_name = scenario_config.get('name', 'Unknown Scenario')
        scenario_type = scenario_config.get('type', 'standard')
        
        # Create scenario-specific conditions
        if scenario_type == 'peak_demand':
            demand_multiplier = 1.4
            weather_impact = 0.9
            equipment_stress = 1.2
        elif scenario_type == 'maintenance_day':
            demand_multiplier = 0.7
            weather_impact = 1.0
            equipment_stress = 0.5
        elif scenario_type == 'weather_event':
            demand_multiplier = 0.8
            weather_impact = 0.6
            equipment_stress = 1.1
        else:
            demand_multiplier = 1.0
            weather_impact = 1.0
            equipment_stress = 1.0
        
        # Simulate scenario performance
        base_productivity = sum(unit['productivity'] for unit in self.equipment_units.values())
        base_energy = sum(self.equipment_types[unit['equipment_type']].energy_consumption 
                         for unit in self.equipment_units.values())
        
        scenario_productivity = base_productivity * demand_multiplier * weather_impact
        scenario_energy = base_energy * equipment_stress
        scenario_efficiency = scenario_productivity / (scenario_energy + 0.001)
        
        # Calculate scenario-specific metrics
        scenario_results = {
            'productivity': scenario_productivity,
            'energy_consumption': scenario_energy,
            'efficiency': scenario_efficiency,
            'utilization': min(1.0, 0.75 * demand_multiplier),
            'downtime_risk': max(0, equipment_stress - 1.0) * 0.1,
            'maintenance_need': equipment_stress * 1.5
        }
        
        self.scenario_results[scenario_name] = scenario_results
        
        return scenario_results

print('Integrated Digital Twin System defined successfully')
```

### Initialize Digital Twin System

```python
# Create digital twin system
digital_twin = IntegratedDigitalTwinSystem(equipment_types)

print('=== DIGITAL TWIN SYSTEM INITIALIZED ===\n')
print(f'Equipment units: {len(digital_twin.equipment_units)}')
print(f'IoT sensors: {len(digital_twin.iot_sensors)}')
print(f'Prediction models: {len(digital_twin.prediction_models)}')
print(f'System uptime: {digital_twin.system_uptime:.2%}')
print(f'Synchronization accuracy: {digital_twin.synchronization_accuracy:.2%}')
print(f'Prediction accuracy: {digital_twin.prediction_accuracy:.2%}\n')

# Display equipment configuration
equipment_summary = {}
for unit in digital_twin.equipment_units.values():
    eq_type = unit['equipment_type']
    if eq_type not in equipment_summary:
        equipment_summary[eq_type] = {'count': 0, 'avg_productivity': 0, 'avg_reliability': 0}
    
    equipment_summary[eq_type]['count'] += 1
    equipment_summary[eq_type]['avg_productivity'] += unit['productivity']
    equipment_summary[eq_type]['avg_reliability'] += unit['reliability']

print('Equipment Fleet Summary:')
for eq_type, summary in equipment_summary.items():
    count = summary['count']
    avg_productivity = summary['avg_productivity'] / count
    avg_reliability = summary['avg_reliability'] / count
    print(f'  {eq_type}: {count} units, Avg Productivity: {avg_productivity:.1f} cph, Avg Reliability: {avg_reliability:.2f}')
```

### Run Real-Time Digital Twin Simulation

```python
# Run 24-hour real-time simulation
print('=== REAL-TIME DIGITAL TWIN SIMULATION ===\n')

simulation_hours = 24
current_time = datetime.now().replace(microsecond=0, second=0, minute=0)

# Store simulation results
time_series_data = {
    'timestamps': [],
    'total_productivity': [],
    'total_energy': [],
    'avg_utilization': [],
    'active_units': [],
    'maintenance_alerts': [],
    'system_efficiency': []
    'prediction_confidence': []
}

for hour in range(simulation_hours):
    # Update current time
    simulation_time = current_time + timedelta(hours=hour)
    
    # Collect sensor data
    sensor_data = digital_twin.collect_sensor_data(simulation_time)
    
    # Update digital twin state
    twin_state = digital_twin.update_digital_twin_state(sensor_data, simulation_time)
    
    # Store time series data
    time_series_data['timestamps'].append(simulation_time)
    time_series_data['total_productivity'].append(twin_state.operational_kpis['total_productivity'])
    time_series_data['total_energy'].append(twin_state.operational_kpis['total_energy_consumption'])
    time_series_data['avg_utilization'].append(twin_state.operational_kpis['average_utilization'])
    time_series_data['active_units'].append(twin_state.operational_kpis['active_equipment_count'])
    time_series_data['maintenance_alerts'].append(twin_state.operational_kpis['maintenance_alerts'])
    time_series_data['system_efficiency'].append(twin_state.operational_kpis['system_efficiency'])
    time_series_data['prediction_confidence'].append(twin_state.prediction_confidence)
    
    # Progress update
    if (hour + 1) % 6 == 0:
        print(f'Hour {hour + 1}: Productivity: {twin_state.operational_kpis["total_productivity"]:.0f} cph, '
              f'Efficiency: {twin_state.operational_kpis["system_efficiency"]:.1f}, '
              f'Maintenance Alerts: {twin_state.operational_kpis["maintenance_alerts"]}')

print(f'\nSimulation completed: {simulation_hours} hours')
print(f'Total digital twin states: {len(digital_twin.digital_twin_states)}')
print(f'Total sensor readings processed: {len(sensor_data) * simulation_hours:,}\n')

# Calculate simulation statistics
avg_productivity = np.mean(time_series_data['total_productivity'])
avg_efficiency = np.mean(time_series_data['system_efficiency'])
avg_utilization = np.mean(time_series_data['avg_utilization'])
max_maintenance_alerts = max(time_series_data['maintenance_alerts'])

print('=== SIMULATION PERFORMANCE METRICS ===\n')
print(f'Average Productivity: {avg_productivity:.0f} containers/hour')
print(f'Average System Efficiency: {avg_efficiency:.1f} containers/kWh')
print(f'Average Utilization: {avg_utilization:.1%}')
print(f'Peak Maintenance Alerts: {max_maintenance_alerts}')
print(f'System Stability: {digital_twin.system_uptime:.1%} uptime')
```

### Generate Predictive Analytics

```python
# Generate 48-hour predictions
print('=== PREDICTIVE ANALYTICS ===\n')

predictions = digital_twin.generate_predictions(horizon_hours=48)

print('48-Hour Performance Predictions:\n')

for model_name, prediction_values in predictions.items():
    model_info = digital_twin.prediction_models[model_name]
    
    print(f'{model_name.upper()} Model:')
    print(f'  Model Type: {model_info["model_type"]}')
    print(f'  Accuracy: {model_info["accuracy"]:.1%}')
    print(f'  Update Frequency: {model_info["update_frequency"]}')
    
    # Calculate prediction statistics
    avg_prediction = np.mean(prediction_values)
    max_prediction = max(prediction_values)
    min_prediction = min(prediction_values)
    trend = 'Increasing' if prediction_values[-1] > prediction_values[0] else 'Decreasing'
    
    print(f'  Average Prediction: {avg_prediction:.2f}')
    print(f'  Range: {min_prediction:.2f} - {max_prediction:.2f}')
    print(f'  Trend: {trend}\n')

# Prediction accuracy analysis
print('=== PREDICTION ACCURACY ANALYSIS ===\n')
overall_accuracy = np.mean([model['accuracy'] for model in digital_twin.prediction_models.values()])
print(f'Overall Prediction Accuracy: {overall_accuracy:.1%}')
print(f'Synchronization Accuracy: {digital_twin.synchronization_accuracy:.1%}')
print(f'Digital Twin Confidence: {digital_twin.digital_twin_states[-1].prediction_confidence:.1%}\n')

# Identify critical predictions
print('=== CRITICAL PREDICTIONS ===\n')
if predictions.get('maintenance'):
    maintenance_predictions = predictions['maintenance']
    peak_maintenance = max(maintenance_predictions)
    peak_hour = maintenance_predictions.index(peak_maintenance)
    
    print(f'Peak Maintenance Need: Hour {peak_hour + 1} ({peak_maintenance:.1f} alerts)')
    if peak_maintenance > 5:
        print('⚠️  HIGH MAINTENANCE ALERT - Schedule preventive maintenance')

if predictions.get('failure'):
    failure_predictions = predictions['failure']
    max_failure_risk = max(failure_predictions)
    if max_failure_risk > 0.1:
        risk_hour = failure_predictions.index(max_failure_risk)
        print(f'⚠️  ELEVATED FAILURE RISK: Hour {risk_hour + 1} ({max_failure_risk:.1%} probability)')

if predictions.get('productivity'):
    productivity_predictions = predictions['productivity']
    min_productivity = min(productivity_predictions)
    if min_productivity < avg_productivity * 0.8:
        low_hour = productivity_predictions.index(min_productivity)
        print(f'⚠️  LOW PRODUCTIVITY: Hour {low_hour + 1} ({min_productivity:.0f} cph)')
```

### Scenario Analysis and What-If Testing

```python
# Define test scenarios
scenarios = [
    {
        'name': 'Normal Operations',
        'type': 'standard',
        'description': 'Standard operating conditions'
    },
    {
        'name': 'Peak Demand Surge',
        'type': 'peak_demand',
        'description': '40% increase in container volume'
    },
    {
        'name': 'Scheduled Maintenance',
        'type': 'maintenance_day',
        'description': '30% equipment under maintenance'
    },
    {
        'name': 'Severe Weather Event',
        'type': 'weather_event',
        'description': 'Poor weather conditions affecting operations'
    },
    {
        'name': 'Equipment Failure',
        'type': 'equipment_failure',
        'description': 'Multiple equipment units offline'
    }
]

print('=== SCENARIO ANALYSIS ===\n')

# Run scenario analysis
scenario_results = {}
baseline_results = None

for scenario in scenarios:
    print(f'--- {scenario["name"]} ---')
    print(f'Description: {scenario["description"]}\n')
    
    # Run scenario
    results = digital_twin.run_scenario_analysis(scenario)
    scenario_results[scenario['name']] = results
    
    # Store baseline for comparison
    if scenario['name'] == 'Normal Operations':
        baseline_results = results
    
    print(f'Scenario Results:')
    print(f'  Productivity: {results["productivity"]:.0f} containers/hour')
    print(f'  Energy Consumption: {results["energy_consumption"]:.1f} kWh')
    print(f'  System Efficiency: {results["efficiency"]:.1f} containers/kWh')
    print(f'  Utilization: {results["utilization"]:.1%}')
    print(f'  Downtime Risk: {results["downtime_risk"]:.1%}')
    print(f'  Maintenance Need: {results["maintenance_need"]:.1f}\n')
    
    # Compare with baseline if available
    if baseline_results and scenario['name'] != 'Normal Operations':
        productivity_change = (results['productivity'] / baseline_results['productivity'] - 1) * 100
        efficiency_change = (results['efficiency'] / baseline_results['efficiency'] - 1) * 100
        
        print(f'vs Normal Operations:')
        print(f'  Productivity Change: {productivity_change:+.1f}%')
        print(f'  Efficiency Change: {efficiency_change:+.1f}%')
        
        if productivity_change < -10:
            print('  ⚠️  Significant performance degradation')
        elif productivity_change > 10:
            print('  📈 Performance improvement opportunity')
        
        print('\n' + '-' * 50 + '\n')

# Scenario recommendations
print('=== SCENARIO-BASED RECOMMENDATIONS ===\n')

recommendations = {
    'Peak Demand Surge': [
        'Activate backup equipment units',
        'Implement overtime scheduling',
        'Prioritize high-efficiency equipment',
        'Consider temporary equipment rental'
    ],
    'Scheduled Maintenance': [
        'Plan maintenance during low-demand periods',
        'Cross-train staff for equipment flexibility',
        'Optimize equipment routing',
        'Increase preventive maintenance frequency'
    ],
    'Severe Weather Event': [
        'Implement weather-specific operating procedures',
        'Reduce operating speeds for safety',
        'Increase inspection frequency',
        'Prepare contingency equipment deployment'
    ],
    'Equipment Failure': [
        'Activate rapid response protocols',
        'Deploy backup equipment immediately',
        'Implement load balancing procedures',
        'Schedule emergency maintenance assessment'
    ]
}

for scenario_name, scenario_recommendations in recommendations.items():
    if scenario_name in scenario_results:
        print(f'{scenario_name}:')
        for i, rec in enumerate(scenario_recommendations, 1):
            print(f'  {i}. {rec}')
        print()
```
