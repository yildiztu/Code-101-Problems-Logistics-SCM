### Value-Aligned Ethical Optimization System

```python
class ValueAlignedEthicalOptimizer:
    def __init__(self, equipment_types: Dict[str, EquipmentType]):
        self.equipment_types = equipment_types
        self.stakeholders = []
        self.ethical_constraints = []
        self.optimization_history = []
        self.constitutional_principles = []
        
        # Ethical optimization parameters
        self.ethical_weight = 0.4  # Weight given to ethical considerations
        self.economic_weight = 0.3  # Weight given to economic efficiency
        self.social_weight = 0.3  # Weight given to social impact
        
        self._initialize_stakeholders()
        self._define_ethical_constraints()
        self._establish_constitutional_principles()
        
    def _initialize_stakeholders(self):
        """Initialize diverse stakeholder groups"""
        stakeholder_configs = [
            {
                'name': 'Terminal Operator',
                'category': 'Economic',
                'priorities': {'cost_efficiency': 0.4, 'productivity': 0.3, 'flexibility': 0.2, 'maintenance': 0.1},
                'influence': 0.25,
                'utility_function': 'economic_utility',
                'ethical_constraints': ['safety_standards', 'regulatory_compliance']
            },
            {
                'name': 'Port Authority',
                'category': 'Regulatory',
                'priorities': {'safety': 0.4, 'environmental_protection': 0.3, 'compliance': 0.2, 'efficiency': 0.1},
                'influence': 0.20,
                'utility_function': 'regulatory_utility',
                'ethical_constraints': ['environmental_limits', 'safety_standards', 'accessibility']
            },
            {
                'name': 'Local Community',
                'category': 'Social',
                'priorities': {'job_creation': 0.35, 'environmental_quality': 0.3, 'noise_reduction': 0.2, 'community_benefits': 0.15},
                'influence': 0.15,
                'utility_function': 'social_utility',
                'ethical_constraints': ['environmental_justice', 'job_protection', 'community_impact']
            },
            {
                'name': 'Environmental Group',
                'category': 'Environmental',
                'priorities': {'environmental_impact': 0.5, 'sustainability': 0.3, 'carbon_footprint': 0.2},
                'influence': 0.20,
                'utility_function': 'environmental_utility',
                'ethical_constraints': ['environmental_limits', 'sustainability_thresholds', 'carbon_neutral']
            },
            {
                'name': 'Labor Union',
                'category': 'Social',
                'priorities': {'job_security': 0.4, 'safety': 0.3, 'fair_wages': 0.2, 'working_conditions': 0.1},
                'influence': 0.10,
                'utility_function': 'labor_utility',
                'ethical_constraints': ['job_protection', 'safety_standards', 'fair_labor_practices']
            },
            {
                'name': 'Government Agency',
                'category': 'Regulatory',
                'priorities': {'economic_development': 0.3, 'environmental_compliance': 0.3, 'public_safety': 0.2, 'tax_revenue': 0.2},
                'influence': 0.10,
                'utility_function': 'government_utility',
                'ethical_constraints': ['regulatory_compliance', 'public_interest', 'economic_viability']
            }
        ]
        
        for config in stakeholder_configs:
            stakeholder = Stakeholder(
                name=config['name'],
                category=config['category'],
                priorities=config['priorities'],
                influence=config['influence'],
                utility_function=config['utility_function'],
                ethical_constraints=config['ethical_constraints']
            )
            self.stakeholders.append(stakeholder)
        
        print(f'Initialized {len(self.stakeholders)} stakeholder groups:')
        for stakeholder in self.stakeholders:
            print(f'  {stakeholder.name} - {stakeholder.category} (Influence: {stakeholder.influence:.2f})')
    
    def _define_ethical_constraints(self):
        """Define comprehensive ethical constraints"""
        constraints = [
            EthicalConstraint(
                name='Environmental Impact Limits',
                category='Environmental',
                description='Total environmental impact must not exceed regulatory thresholds',
                weight=0.25,
                threshold=0.7,  # Maximum average impact score
                check_function='check_environmental_impact'
            ),
            EthicalConstraint(
                name='Labor Safety Standards',
                category='Social',
                description='Minimum safety standards must be met for all equipment',
                weight=0.30,
                threshold=0.8,  # Minimum average safety rating
                check_function='check_labor_safety'
            ),
            EthicalConstraint(
                name='Job Protection Requirements',
                category='Social',
                description='Maintain minimum employment levels and job quality',
                weight=0.20,
                threshold=50,  # Minimum total jobs created
                check_function='check_job_protection'
            ),
            EthicalConstraint(
                name='Cost Fairness Distribution',
                category='Economic',
                description='Ensure fair distribution of costs and benefits',
                weight=0.15,
                threshold=0.6,  # Minimum fairness score
                check_function='check_cost_fairness'
            ),
            EthicalConstraint(
                name='Community Impact Balance',
                category='Social',
                description='Ensure positive net impact on local community',
                weight=0.10,
                threshold=0.5,  # Minimum positive community impact
                check_function='check_community_impact'
            )
        ]
        
        self.ethical_constraints = constraints
        print(f'\nDefined {len(self.ethical_constraints)} ethical constraints:')
        for constraint in self.ethical_constraints:
            print(f'  {constraint.name} ({constraint.category}) - Weight: {constraint.weight:.2f}')
    
    def _establish_constitutional_principles(self):
        """Establish constitutional AI principles for ethical decision making"""
        principles = [
            {
                'name': 'Do No Harm',
                'description': 'Avoid actions that cause significant harm to stakeholders',
                'priority': 1,
                'application': 'veto_power'
            },
            {
                'name': 'Beneficence',
                'description': 'Actively promote stakeholder wellbeing',
                'priority': 2,
                'application': 'positive_weight'
            },
            {
                'name': 'Justice',
                'description': 'Ensure fair distribution of benefits and burdens',
                'priority': 2,
                'application': 'balancing_factor'
            },
            {
                'name': 'Autonomy',
                'description': 'Respect stakeholder decision-making autonomy',
                'priority': 3,
                'application': 'consultation_requirement'
            },
            {
                'name': 'Sustainability',
                'description': 'Ensure long-term environmental and economic sustainability',
                'priority': 2,
                'application': 'constraint_threshold'
            }
        ]
        
        self.constitutional_principles = principles
        print(f'\nEstablished {len(self.constitutional_principles)} constitutional principles:')
        for principle in self.constitutional_principles:
            print(f'  {principle["name"]} (Priority: {principle["priority"]})')
    
    def check_environmental_impact(self, equipment_mix: Dict[str, int]) -> float:
        """Check environmental impact constraint compliance"""
        total_units = sum(equipment_mix.values())
        if total_units == 0:
            return 0.0
        
        # Calculate weighted average environmental impact
        total_impact = 0
        for eq_type, count in equipment_mix.items():
            impact = self.equipment_types[eq_type].environmental_impact
            total_impact += impact * count
        
        avg_impact = total_impact / total_units
        
        # Compliance score (lower impact is better)
        threshold = self.ethical_constraints[0].threshold
        if avg_impact <= threshold:
            compliance_score = 1.0
        else:
            # Penalize excess impact
            excess = avg_impact - threshold
            compliance_score = max(0.0, 1.0 - (excess / (1.0 - threshold)))
        
        return compliance_score
    
    def check_labor_safety(self, equipment_mix: Dict[str, int]) -> float:
        """Check labor safety constraint compliance"""
        total_units = sum(equipment_mix.values())
        if total_units == 0:
            return 0.0
        
        # Calculate weighted average safety rating
        total_safety = 0
        for eq_type, count in equipment_mix.items():
            safety = self.equipment_types[eq_type].safety_rating
            total_safety += safety * count
        
        avg_safety = total_safety / total_units
        
        # Compliance score (higher safety is better)
        threshold = self.ethical_constraints[1].threshold
        if avg_safety >= threshold:
            compliance_score = 1.0
        else:
            # Penalize insufficient safety
            deficit = threshold - avg_safety
            compliance_score = max(0.0, 1.0 - (deficit / threshold))
        
        return compliance_score
    
    def check_job_protection(self, equipment_mix: Dict[str, int]) -> float:
        """Check job protection constraint compliance"""
        total_jobs = 0
        for eq_type, count in equipment_mix.items():
            jobs = self.equipment_types[eq_type].job_creation_factor * count
            total_jobs += jobs
        
        threshold = self.ethical_constraints[2].threshold
        if total_jobs >= threshold:
            compliance_score = 1.0
        else:
            # Penalize insufficient job creation
            deficit = threshold - total_jobs
            compliance_score = max(0.0, 1.0 - (deficit / threshold))
        
        return compliance_score
    
    def check_cost_fairness(self, equipment_mix: Dict[str, int]) -> float:
        """Check cost fairness constraint compliance"""
        total_units = sum(equipment_mix.values())
        if total_units == 0:
            return 0.0
        
        # Calculate cost distribution fairness (prefer balanced distribution)
        ideal_distribution = total_units / len(equipment_mix)
        fairness_score = 1.0
        
        for eq_type, count in equipment_mix.items():
            deviation = abs(count - ideal_distribution)
            fairness_score -= (deviation / ideal_distribution) * 0.2
        
        fairness_score = max(0.0, fairness_score)
        
        threshold = self.ethical_constraints[3].threshold
        if fairness_score >= threshold:
            compliance_score = 1.0
        else:
            deficit = threshold - fairness_score
            compliance_score = max(0.0, 1.0 - (deficit / threshold))
        
        return compliance_score
    
    def check_community_impact(self, equipment_mix: Dict[str, int]) -> float:
        """Check community impact constraint compliance"""
        total_units = sum(equipment_mix.values())
        if total_units == 0:
            return 0.0
        
        # Calculate weighted average community impact
        total_impact = 0
        for eq_type, count in equipment_mix.items():
            impact = self.equipment_types[eq_type].community_impact
            total_impact += impact * count
        
        avg_impact = total_impact / total_units
        
        threshold = self.ethical_constraints[4].threshold
        if avg_impact >= threshold:
            compliance_score = 1.0
        else:
            deficit = threshold - avg_impact
            compliance_score = max(0.0, 1.0 - (deficit / threshold))
        
        return compliance_score
    
    def calculate_stakeholder_utility(self, equipment_mix: Dict[str, int], stakeholder: Stakeholder) -> float:
        """Calculate utility for a specific stakeholder"""
        utility = 0.0
        
        # Calculate basic metrics
        total_cost = sum(equipment_mix.get(eq_type, 0) * self.equipment_types[eq_type].capital_cost 
                       for eq_type in self.equipment_types.keys())
        total_productivity = sum(equipment_mix.get(eq_type, 0) * self.equipment_types[eq_type].productivity 
                           for eq_type in self.equipment_types.keys())
        total_units = sum(equipment_mix.values())
        
        if total_units == 0:
            return 0.0
        
        # Calculate weighted averages
        avg_flexibility = sum(equipment_mix.get(eq_type, 0) * self.equipment_types[eq_type].flexibility 
                            for eq_type in self.equipment_types.keys()) / total_units
        avg_environmental_impact = sum(equipment_mix.get(eq_type, 0) * self.equipment_types[eq_type].environmental_impact 
                                  for eq_type in self.equipment_types.keys()) / total_units
        avg_safety = sum(equipment_mix.get(eq_type, 0) * self.equipment_types[eq_type].safety_rating 
                     for eq_type in self.equipment_types.keys()) / total_units
        total_jobs = sum(equipment_mix.get(eq_type, 0) * self.equipment_types[eq_type].job_creation_factor 
                      for eq_type in self.equipment_types.keys())
        avg_community_impact = sum(equipment_mix.get(eq_type, 0) * self.equipment_types[eq_type].community_impact 
                                for eq_type in self.equipment_types.keys()) / total_units
        
        # Calculate utility based on stakeholder priorities
        for priority, weight in stakeholder.priorities.items():
            if priority == 'cost_efficiency':
                # Lower cost is better (inverse)
                cost_efficiency = 1.0 / (1.0 + total_cost / 100)  # Normalize
                utility += weight * cost_efficiency
            elif priority == 'productivity':
                productivity_score = total_productivity / 1000  # Normalize
                utility += weight * productivity_score
            elif priority == 'flexibility':
                utility += weight * avg_flexibility
            elif priority == 'maintenance':
                # Lower maintenance is better
                total_maintenance = sum(equipment_mix.get(eq_type, 0) * self.equipment_types[eq_type].maintenance 
                                   for eq_type in self.equipment_types.keys())
                maintenance_score = 1.0 / (1.0 + total_maintenance / 10)
                utility += weight * maintenance_score
            elif priority == 'safety':
                utility += weight * avg_safety
            elif priority == 'environmental_protection' or priority == 'environmental_impact':
                # Lower environmental impact is better
                environmental_score = 1.0 - avg_environmental_impact
                utility += weight * environmental_score
            elif priority == 'job_creation' or priority == 'job_security':
                job_score = min(1.0, total_jobs / 100)  # Normalize
                utility += weight * job_score
            elif priority == 'community_benefits' or priority == 'community_impact':
                utility += weight * avg_community_impact
            elif priority == 'compliance':
                # Higher compliance is better
                compliance_scores = [self.check_environmental_impact(equipment_mix),
                                   self.check_labor_safety(equipment_mix),
                                   self.check_job_protection(equipment_mix)]
                avg_compliance = np.mean(compliance_scores)
                utility += weight * avg_compliance
            elif priority == 'efficiency':
                efficiency_score = total_productivity / (total_cost + 1)  # Avoid division by zero
                utility += weight * min(1.0, efficiency_score / 50)  # Normalize
            # Add more priority calculations as needed
        
        return utility
    
    def calculate_overall_utility(self, equipment_mix: Dict[str, int]) -> float:
        """Calculate weighted overall utility across all stakeholders"""
        total_utility = 0.0
        total_influence = sum(stakeholder.influence for stakeholder in self.stakeholders)
        
        for stakeholder in self.stakeholders:
            stakeholder_utility = self.calculate_stakeholder_utility(equipment_mix, stakeholder)
            weighted_utility = stakeholder_utility * (stakeholder.influence / total_influence)
            total_utility += weighted_utility
        
        return total_utility
    
    def calculate_ethical_compliance(self, equipment_mix: Dict[str, int]) -> Dict[str, float]:
        """Calculate compliance scores for all ethical constraints"""
        compliance_scores = {}
        
        for constraint in self.ethical_constraints:
            # Call the appropriate check function
            check_function = getattr(self, constraint.check_function)
            score = check_function(equipment_mix)
            compliance_scores[constraint.name] = score
        
        return compliance_scores
    
    def apply_constitutional_principles(self, solution: EthicalSolution) -> EthicalSolution:
        """Apply constitutional AI principles to validate or modify solution"""
        # Check 'Do No Harm' principle (veto power)
        min_compliance = min(solution.ethical_compliance.values())
        if min_compliance < 0.3:  # Significant harm threshold
            # Modify solution to reduce harm
            solution.explanation.append("Applied 'Do No Harm' principle - modified to reduce harm")
            # This is simplified - in practice, would use more sophisticated modification
        
        return solution
    
    def generate_ethical_solution(self, equipment_mix: Dict[str, int]) -> EthicalSolution:
        """Generate comprehensive ethical solution analysis"""
        # Calculate basic metrics
        total_cost = sum(equipment_mix.get(eq_type, 0) * self.equipment_types[eq_type].capital_cost 
                       for eq_type in self.equipment_types.keys())
        total_productivity = sum(equipment_mix.get(eq_type, 0) * self.equipment_types[eq_type].productivity 
                           for eq_type in self.equipment_types.keys())
        total_units = sum(equipment_mix.values())
        
        if total_units == 0:
            return EthicalSolution(
                equipment_config=equipment_mix,
                overall_utility=0.0,
                stakeholder_utilities={},
                ethical_compliance={},
                total_cost=0.0,
                total_productivity=0.0,
                environmental_impact=0.0,
                social_impact=0.0,
                explanation=["No equipment selected - invalid solution"]
            )
        
        # Calculate stakeholder utilities
        stakeholder_utilities = {}
        for stakeholder in self.stakeholders:
            utility = self.calculate_stakeholder_utility(equipment_mix, stakeholder)
            stakeholder_utilities[stakeholder.name] = utility
        
        # Calculate overall utility
        overall_utility = self.calculate_overall_utility(equipment_mix)
        
        # Calculate ethical compliance
        ethical_compliance = self.calculate_ethical_compliance(equipment_mix)
        
        # Calculate environmental and social impact
        avg_environmental_impact = sum(equipment_mix.get(eq_type, 0) * self.equipment_types[eq_type].environmental_impact 
                                    for eq_type in self.equipment_types.keys()) / total_units
        total_jobs = sum(equipment_mix.get(eq_type, 0) * self.equipment_types[eq_type].job_creation_factor 
                      for eq_type in self.equipment_types.keys())
        avg_community_impact = sum(equipment_mix.get(eq_type, 0) * self.equipment_types[eq_type].community_impact 
                                for eq_type in self.equipment_types.keys()) / total_units
        
        # Generate explanation
        explanation = []
        explanation.append(f"Selected configuration: {equipment_mix}")
        explanation.append(f"Total cost: ${total_cost:.1f}M, Productivity: {total_productivity:.0f} cph")
        explanation.append(f"Overall utility: {overall_utility:.3f}")
        
        # Add ethical constraint analysis
        min_compliance = min(ethical_compliance.values())
        if min_compliance >= 0.9:
            explanation.append("Excellent ethical compliance across all constraints")
        elif min_compliance >= 0.7:
            explanation.append("Good ethical compliance with minor trade-offs")
        else:
            explanation.append("Some ethical constraints require attention")
        
        # Create solution
        solution = EthicalSolution(
            equipment_config=equipment_mix,
            overall_utility=overall_utility,
            stakeholder_utilities=stakeholder_utilities,
            ethical_compliance=ethical_compliance,
            total_cost=total_cost,
            total_productivity=total_productivity,
            environmental_impact=avg_environmental_impact,
            social_impact=total_jobs,
            explanation=explanation
        )
        
        # Apply constitutional principles
        solution = self.apply_constitutional_principles(solution)
        
        return solution
    
    def optimize_equipment_selection_ethical(self, budget_limit: float = 200, 
                                            min_productivity: float = 500) -> EthicalSolution:
        """Optimize equipment selection with ethical constraints"""
        print(f'=== VALUE-ALIGNED ETHICAL OPTIMIZATION ===\n')
        print(f'Budget limit: ${budget_limit}M')
        print(f'Minimum productivity: {min_productivity} containers/hour\n')
        
        # Generate candidate solutions
        candidates = self._generate_ethical_candidates(budget_limit)
        
        best_solution = None
        best_score = -float('inf')
        
        print(f'Evaluating {len(candidates)} candidate solutions...\n')
        
        for i, candidate in enumerate(candidates):
            solution = self.generate_ethical_solution(candidate)
            
            # Check basic feasibility
            if solution.total_cost > budget_limit or solution.total_productivity < min_productivity:
                continue
            
            # Calculate comprehensive score
            # Combine utility, ethical compliance, and efficiency
            min_ethical_compliance = min(solution.ethical_compliance.values())
            
            # Apply constitutional principles
            if min_ethical_compliance < 0.3:  # 'Do No Harm' violation
                continue  # Skip solutions that cause significant harm
            
            # Comprehensive scoring
            utility_score = solution.overall_utility * self.economic_weight
            ethical_score = min_ethical_compliance * self.ethical_weight
            efficiency_score = (solution.total_productivity / solution.total_cost) * self.social_weight
            
            final_score = utility_score + ethical_score + efficiency_score
            
            print(f'Candidate {i+1}: {candidate}')
            print(f'  Utility: {solution.overall_utility:.3f}, Ethics: {min_ethical_compliance:.3f}, Efficiency: {solution.total_productivity/solution.total_cost:.2f}')
            print(f'  Final Score: {final_score:.3f}\n')
            
            if final_score > best_score:
                best_score = final_score
                best_solution = solution
        
        return best_solution
    
    def _generate_ethical_candidates(self, budget_limit: float) -> List[Dict[str, int]]:
        """Generate diverse candidate solutions for ethical optimization"""
        candidates = []
        
        # Generate balanced configurations
        base_configs = [
            {'RTG': 12, 'RMG': 8, 'Straddle': 10},  # Balanced
            {'RTG': 15, 'RMG': 6, 'Straddle': 9},   # RTG-focused
            {'RTG': 8, 'RMG': 12, 'Straddle': 10},  # RMG-focused
            {'RTG': 10, 'RMG': 7, 'Straddle': 13},  # Straddle-focused
            {'RTG': 14, 'RMG': 10, 'Straddle': 6}, # High efficiency
            {'RTG': 6, 'RMG': 8, 'Straddle': 16},  # High employment
            {'RTG': 11, 'RMG': 9, 'Straddle': 10},  # Ethical balance
            {'RTG': 9, 'RMG': 11, 'Straddle': 10},  # Environmental focus
        ]
        
        # Scale configurations to budget
        for config in base_configs:
            total_cost = sum(config[eq_type] * self.equipment_types[eq_type].capital_cost 
                           for eq_type in config.keys())
            
            if total_cost <= budget_limit * 1.1:  # Allow some budget flexibility
                # Scale to fit budget exactly
                scale_factor = budget_limit / total_cost
                scaled_config = {eq_type: max(1, int(count * scale_factor)) 
                                for eq_type, count in config.items()}
                candidates.append(scaled_config)
        
        return candidates

print('Value-Aligned Ethical Optimization System defined successfully')
```
