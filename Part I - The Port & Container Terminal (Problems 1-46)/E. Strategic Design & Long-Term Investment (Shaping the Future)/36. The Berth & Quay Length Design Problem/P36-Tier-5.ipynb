```python
# Import required packages for Digital Twin System
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Dict, List, Tuple, Optional, Any
import random
import time
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Set up visualization style
plt.style.use('default')
sns.set_palette("husl")
```

```python
# Define data structures for Digital Twin System
from dataclasses import dataclass
from typing import List, Dict, Tuple, Optional
from enum import Enum

class VesselStatus(Enum):
    APPROACHING = "approaching"
    AT_ANCHOR = "at_anchor"
    BERTHING = "berthing"
    BERTHED = "berthed"
    DEPARTING = "departing"
    DEPARTED = "departed"

@dataclass
class Vessel:
    """Represents a vessel in the digital twin system"""
    id: str
    name: str
    length: float  # meters
    draft: float   # meters
    cargo_capacity: float  # TEU
    current_status: VesselStatus
    eta: datetime
    current_position: Tuple[float, float]  # (lat, lon)
    assigned_berth: Optional[int] = None
    cargo_type: str = "General"

@dataclass
class Berth:
    """Represents a berth in the digital twin system"""
    id: int
    length: float  # meters
    depth: float   # meters
    current_vessel: Optional[Vessel] = None
    utilization_rate: float = 0.0
    maintenance_status: str = "Operational"
    structural_health: float = 1.0  # 0-1 scale

@dataclass
class SensorData:
    """Represents sensor data from physical assets"""
    sensor_id: str
    sensor_type: str
    timestamp: datetime
    value: float
    unit: str
    location: str

@dataclass
class WeatherData:
    """Represents environmental conditions"""
    timestamp: datetime
    wind_speed: float  # knots
    wind_direction: float  # degrees
    wave_height: float  # meters
    visibility: float  # nautical miles
    temperature: float  # Celsius

@dataclass
class PredictionResult:
    """Represents a prediction from analytics engine"""
    prediction_type: str
    target_time: datetime
    predicted_value: float
    confidence: float  # 0-1 scale
    model_version: str
```

```python
# Digital Twin Architecture Components
class DigitalTwinArchitecture:
    """Five-layer digital twin architecture for port infrastructure"""
    
    def __init__(self, quay_length: float, num_berths: int):
        self.quay_length = quay_length
        self.num_berths = num_berths
        
        # Layer 1: Physical Asset Layer
        self.berths = self._initialize_berths()
        self.sensors = self._initialize_sensors()
        
        # Layer 2: Operational Data Layer
        self.vessels = []
        self.ais_data = []
        self.port_management_data = {}
        
        # Layer 3: Analytics Layer
        self.predictions = []
        self.ml_models = self._initialize_ml_models()
        
        # Layer 4: Simulation Layer
        self.simulation_state = "running"
        self.simulation_time = datetime.now()
        
        # Layer 5: Decision Support Layer
        self.optimization_engine = OptimizationEngine()
        self.recommendations = []
        
        # System metrics
        self.kpi_monitor = KPIMonitor()
        
    def _initialize_berths(self) -> List[Berth]:
        """Initialize berths based on Mediterranean Logistics Hub example"""
        berths = []
        
        # Create 8 berths with varying sizes (total 3.2km)
        berth_configs = [
            (400, 16), (380, 15), (420, 17), (390, 16),  # Berths 1-4
            (410, 16), (400, 16), (390, 15), (410, 17)   # Berths 5-8
        ]
        
        for i, (length, depth) in enumerate(berth_configs):
            berth = Berth(
                id=i+1,
                length=length,
                depth=depth,
                structural_health=random.uniform(0.85, 1.0)
            )
            berths.append(berth)
        
        return berths
    
    def _initialize_sensors(self) -> Dict[str, List[SensorData]]:
        """Initialize sensor infrastructure based on source example"""
        sensors = {
            "structural_health": [],
            "weather": [],
            "water_level": [],
            "vessel_detection": [],
            "cargo_handling": []
            "yard_operations": []
            "environmental": []
        }
        
        # Generate initial sensor data
        current_time = datetime.now()
        
        # Structural health sensors (240 sensors)
        for i in range(240):
            sensor = SensorData(
                sensor_id=f"SH_{i:03d}",
                sensor_type="structural_health",
                timestamp=current_time,
                value=random.uniform(0.85, 1.0),
                unit="health_index",
                location=f"Berth_{(i%8)+1}_Section_{(i%30)+1}"
            )
            sensors["structural_health"].append(sensor)
        
        # Weather stations (15 stations)
        for i in range(15):
            sensor = SensorData(
                sensor_id=f"WS_{i:02d}",
                sensor_type="weather",
                timestamp=current_time,
                value=random.uniform(5, 25),  # wind speed
                unit="knots",
                location=f"Station_{i+1}"
            )
            sensors["weather"].append(sensor)
        
        # Water level monitors (32 monitors)
        for i in range(32):
            sensor = SensorData(
                sensor_id=f"WL_{i:02d}",
                sensor_type="water_level",
                timestamp=current_time,
                value=random.uniform(12.5, 13.5),  # water depth
                unit="meters",
                location=f"Monitor_{i+1}"
            )
            sensors["water_level"].append(sensor)
        
        # Vessel detection radars (18 radars)
        for i in range(18):
            sensor = SensorData(
                sensor_id=f"VR_{i:02d}",
                sensor_type="vessel_detection",
                timestamp=current_time,
                value=random.uniform(0, 10),  # vessels detected
                unit="count",
                location=f"Radar_{i+1}"
            )
            sensors["vessel_detection"].append(sensor)
        
        return sensors
    
    def _initialize_ml_models(self) -> Dict[str, Any]:
        """Initialize machine learning models for predictions"""
        return {
            "vessel_arrival_prediction": {
                "model_type": "LSTM",
                "accuracy": 0.94,
                "prediction_horizon": "48_hours",
                "last_training": datetime.now() - timedelta(days=7)
            },
            "cargo_volume_forecasting": {
                "model_type": "Random_Forest",
                "accuracy": 0.87,
                "prediction_horizon": "7_days",
                "last_training": datetime.now() - timedelta(days=3)
            },
            "equipment_failure_prediction": {
                "model_type": "XGBoost",
                "accuracy": 0.91,
                "prediction_horizon": "72_hours",
                "last_training": datetime.now() - timedelta(days=1)
            }
        }

class OptimizationEngine:
    """Real-time optimization engine for berth allocation"""
    
    def __init__(self):
        self.constraint_categories = 15
        self.optimization_time_limit = 45  # seconds
        self.performance_metrics = {
            "avg_decision_time": 0.0,
            "solution_quality": 0.0,
            "constraint_satisfaction": 0.0
        }
    
    def optimize_berth_allocation(self, vessels: List[Vessel], berths: List[Berth], 
                                 constraints: Dict) -> Dict:
        """Optimize berth allocation considering multiple constraints"""
        start_time = time.time()
        
        # Simplified optimization for demonstration
        allocation = {}
        
        for vessel in vessels:
            if vessel.current_status in [VesselStatus.APPROACHING, VesselStatus.AT_ANCHOR]:
                # Find suitable berth
                suitable_berths = [
                    b for b in berths 
                    if b.length >= vessel.length and 
                       b.depth >= vessel.draft and 
                       b.current_vessel is None and
                       b.maintenance_status == "Operational"
                ]
                
                if suitable_berths:
                    # Select berth with lowest utilization
                    best_berth = min(suitable_berths, key=lambda b: b.utilization_rate)
                    allocation[vessel.id] = best_berth.id
                    best_berth.current_vessel = vessel
                    best_berth.utilization_rate = min(1.0, best_berth.utilization_rate + 0.1)
        
        decision_time = time.time() - start_time
        
        return {
            "allocation": allocation,
            "decision_time": decision_time,
            "vessels_assigned": len(allocation),
            "optimization_score": len(allocation) / max(len(vessels), 1)
        }

class KPIMonitor:
    """Real-time KPI monitoring system"""
    
    def __init__(self):
        self.kpis = {
            "berth_utilization": 0.0,
            "vessel_waiting_time": 0.0,
            "cargo_throughput": 0.0,
            "equipment_efficiency": 0.0,
            "safety_incidents": 0,
            "environmental_compliance": 1.0,
            "infrastructure_roi": 0.0
        }
        self.historical_data = []
    
    def update_kpis(self, berth_data: Dict, vessel_data: Dict, 
                   cargo_data: Dict, timestamp: datetime):
        """Update KPI values based on current data"""
        
        # Calculate berth utilization
        total_berths = len(berth_data.get("berths", []))
        occupied_berths = sum(1 for b in berth_data.get("berths", []) 
                           if b.current_vessel is not None)
        self.kpis["berth_utilization"] = occupied_berths / max(total_berths, 1)
        
        # Calculate average waiting time
        waiting_vessels = [v for v in vessel_data.get("vessels", []) 
                         if v.current_status == VesselStatus.AT_ANCHOR]
        if waiting_vessels:
            avg_wait = np.mean([(datetime.now() - v.eta).total_seconds() / 3600 
                               for v in waiting_vessels])
            self.kpis["vessel_waiting_time"] = avg_wait
        
        # Store historical data
        self.historical_data.append({
            "timestamp": timestamp,
            "kpis": self.kpis.copy()
        })
        
        # Keep only last 24 hours of data
        cutoff_time = timestamp - timedelta(hours=24)
        self.historical_data = [d for d in self.historical_data 
                             if d["timestamp"] > cutoff_time]
    
    def get_performance_summary(self) -> Dict:
        """Get current performance summary"""
        return {
            "current_kpis": self.kpis,
            "trend_analysis": self._calculate_trends(),
            "alerts": self._generate_alerts()
        }
    
    def _calculate_trends(self) -> Dict:
        """Calculate KPI trends"""
        if len(self.historical_data) < 2:
            return {}
        
        trends = {}
        recent_data = self.historical_data[-10:]  # Last 10 data points
        
        for kpi_name in self.kpis.keys():
            values = [d["kpis"][kpi_name] for d in recent_data]
            if len(values) > 1:
                trend = (values[-1] - values[0]) / len(values)
                trends[kpi_name] = trend
        
        return trends
    
    def _generate_alerts(self) -> List[str]:
        """Generate alerts based on KPI thresholds"""
        alerts = []
        
        if self.kpis["berth_utilization"] > 0.9:
            alerts.append("High berth utilization - consider expansion")
        
        if self.kpis["vessel_waiting_time"] > 6.0:
            alerts.append("Excessive vessel waiting times")
        
        if self.kpis["environmental_compliance"] < 0.8:
            alerts.append("Environmental compliance issues detected")
        
        return alerts
```

```python
# System Integration Components
class VesselTrafficManagementSystem:
    """Integration with vessel traffic management and AIS data"""
    
    def __init__(self):
        self.ais_range = 50  # nautical miles
        self.vessels_tracked = []
        self.prediction_accuracy = 0.94
    
    def process_ais_data(self, raw_data: List[Dict]) -> List[Vessel]:
        """Process AIS data and update vessel positions"""
        vessels = []
        
        for data in raw_data:
            vessel = Vessel(
                id=data["mmsi"],
                name=data.get("name", "Unknown"),
                length=data.get("length", 200),
                draft=data.get("draft", 10),
                cargo_capacity=data.get("cargo_capacity", 1000),
                current_status=VesselStatus.APPROACHING,
                eta=data.get("eta", datetime.now() + timedelta(hours=6)),
                current_position=(data["lat"], data["lon"]),
                cargo_type=data.get("cargo_type", "General")
            )
            vessels.append(vessel)
        
        return vessels
    
    def predict_vessel_arrivals(self, time_horizon: int) -> List[PredictionResult]:
        """Predict vessel arrivals using ML models"""
        predictions = []
        current_time = datetime.now()
        
        # Generate mock predictions (94% accuracy from source)
        for hour in range(1, time_horizon + 1):
            prediction_time = current_time + timedelta(hours=hour)
            predicted_arrivals = random.poisson(2.5)  # Average 2.5 arrivals per hour
            confidence = 0.94 * (1 - 0.1 * hour / time_horizon)  # Decreasing confidence
            
            prediction = PredictionResult(
                prediction_type="vessel_arrival",
                target_time=prediction_time,
                predicted_value=predicted_arrivals,
                confidence=confidence,
                model_version="LSTM_v2.1"
            )
            predictions.append(prediction)
        
        return predictions

class CargoHandlingEquipmentSystem:
    """Integration with cargo handling equipment coordination"""
    
    def __init__(self):
        self.cranes = 18  # From source example
        self.equipment_status = {}
        self.coordination_efficiency = 0.87
    
    def coordinate_with_berths(self, berth_assignments: Dict) -> Dict:
        """Coordinate crane operations with berth assignments"""
        coordination_plan = {}
        
        for vessel_id, berth_id in berth_assignments.items():
            # Assign cranes based on vessel size and berth characteristics
            required_cranes = random.randint(2, 4)
            available_cranes = [f"QC_{i}" for i in range(1, self.cranes + 1) 
                               if f"QC_{i}" not in self.equipment_status.values()]
            
            if len(available_cranes) >= required_cranes:
                assigned_cranes = available_cranes[:required_cranes]
                coordination_plan[vessel_id] = {
                    "berth_id": berth_id,
                    "assigned_cranes": assigned_cranes,
                    "estimated_handling_time": random.uniform(8, 24),  # hours
                    "efficiency_score": random.uniform(0.8, 0.95)
                }
                
                # Mark cranes as assigned
                for crane in assigned_cranes:
                    self.equipment_status[crane] = vessel_id
        
        return coordination_plan

class YardOperationsSystem:
    """Integration with yard operations and intermodal connections"""
    
    def __init__(self):
        self.yard_capacity = 2400  # TEU (from source)
        self.truck_connections = 150
        self.rail_connections = 12
        self.utilization_rate = 0.75
    
    def synchronize_with_berths(self, berth_operations: List[Dict]) -> Dict:
        """Synchronize yard operations with berth activities"""
        synchronization_plan = {
            "truck_allocation": {},
            "rail_allocation": {},
            "yard_utilization": self.utilization_rate,
            "bottleneck_analysis": []
        }
        
        for operation in berth_operations:
            vessel_id = operation.get("vessel_id")
            cargo_volume = operation.get("cargo_volume", 1000)
            
            # Allocate trucks and rail cars
            required_trucks = min(self.truck_connections, int(cargo_volume / 50))
            required_rail = min(self.rail_connections, max(1, int(cargo_volume / 200)))
            
            synchronization_plan["truck_allocation"][vessel_id] = required_trucks
            synchronization_plan["rail_allocation"][vessel_id] = required_rail
        
        # Analyze bottlenecks
        if self.utilization_rate > 0.85:
            synchronization_plan["bottleneck_analysis"].append("High yard utilization")
        
        if self.truck_connections < 100:
            synchronization_plan["bottleneck_analysis"].append("Limited truck capacity")
        
        return synchronization_plan

class EnvironmentalMonitoringSystem:
    """Integration with environmental monitoring systems"""
    
    def __init__(self):
        self.weather_stations = 15
        self.current_conditions = WeatherData(
            timestamp=datetime.now(),
            wind_speed=12.5,
            wind_direction=225,
            wave_height=1.2,
            visibility=10.0,
            temperature=18.5
        )
    
    def assess_operational_impact(self) -> Dict:
        """Assess environmental impact on operations"""
        impact_assessment = {
            "berthing_restrictions": [],
            "safety_alerts": [],
            "efficiency_impact": 0.0,
            "recommendations": []
        }
        
        # Check wind conditions
        if self.current_conditions.wind_speed > 25:
            impact_assessment["berthing_restrictions"].append("High wind - limited berthing")
            impact_assessment["efficiency_impact"] += 0.2
        
        # Check visibility
        if self.current_conditions.visibility < 1.0:
            impact_assessment["safety_alerts"].append("Low visibility - increased safety protocols")
            impact_assessment["efficiency_impact"] += 0.15
        
        # Check wave height
        if self.current_conditions.wave_height > 2.0:
            impact_assessment["berthing_restrictions"].append("High waves - rough berthing")
            impact_assessment["efficiency_impact"] += 0.25
        
        # Generate recommendations
        if impact_assessment["efficiency_impact"] > 0.1:
            impact_assessment["recommendations"].append("Consider delaying sensitive operations")
        
        return impact_assessment
```

```python
# Predictive Analytics Engine
class PredictiveAnalyticsEngine:
    """Machine learning models for demand forecasting and performance prediction"""
    
    def __init__(self):
        self.models = {
            "vessel_arrival_lstm": {
                "accuracy": 0.94,
                "horizon_hours": 48,
                "last_updated": datetime.now() - timedelta(days=7)
            },
            "cargo_volume_rf": {
                "accuracy": 0.87,
                "horizon_days": 7,
                "last_updated": datetime.now() - timedelta(days=3)
            },
            "equipment_failure_xgb": {
                "accuracy": 0.91,
                "horizon_hours": 72,
                "last_updated": datetime.now() - timedelta(days=1)
            }
        }
        
        self.prediction_cache = []
    
    def predict_vessel_arrivals(self, historical_data: List[Dict], 
                              prediction_horizon: int) -> List[PredictionResult]:
        """Predict vessel arrivals using LSTM model"""
        predictions = []
        current_time = datetime.now()
        
        # Simulate LSTM predictions with 94% accuracy
        for hour in range(1, min(prediction_horizon, 48) + 1):
            prediction_time = current_time + timedelta(hours=hour)
            
            # Base arrival rate with seasonal variation
            base_rate = 2.5
            seasonal_factor = 1.0 + 0.3 * np.sin(2 * np.pi * hour / 24)  # Daily pattern
            predicted_arrivals = max(0, np.random.poisson(base_rate * seasonal_factor))
            
            # Confidence decreases with prediction horizon
            confidence = 0.94 * (1 - 0.3 * hour / 48)
            
            prediction = PredictionResult(
                prediction_type="vessel_arrival",
                target_time=prediction_time,
                predicted_value=predicted_arrivals,
                confidence=confidence,
                model_version="LSTM_v2.1"
            )
            predictions.append(prediction)
        
        self.prediction_cache.extend(predictions)
        return predictions
    
    def predict_cargo_volumes(self, historical_data: List[Dict], 
                             prediction_horizon: int) -> List[PredictionResult]:
        """Predict cargo volumes using Random Forest model"""
        predictions = []
        current_time = datetime.now()
        
        # Simulate Random Forest predictions with 87% accuracy
        for day in range(1, min(prediction_horizon, 7) + 1):
            prediction_time = current_time + timedelta(days=day)
            
            # Base cargo volume with weekly pattern
            base_volume = 15000  # TEU per day
            weekly_factor = 1.0 + 0.2 * np.sin(2 * np.pi * day / 7)  # Weekly pattern
            predicted_volume = max(0, np.random.normal(base_volume * weekly_factor, 2000))
            
            # Confidence decreases with prediction horizon
            confidence = 0.87 * (1 - 0.2 * day / 7)
            
            prediction = PredictionResult(
                prediction_type="cargo_volume",
                target_time=prediction_time,
                predicted_value=predicted_volume,
                confidence=confidence,
                model_version="RF_v3.2"
            )
            predictions.append(prediction)
        
        return predictions
    
    def predict_equipment_failures(self, sensor_data: List[SensorData], 
                                prediction_horizon: int) -> List[PredictionResult]:
        """Predict equipment failures using XGBoost model"""
        predictions = []
        current_time = datetime.now()
        
        # Analyze sensor data for failure patterns
        structural_sensors = [s for s in sensor_data if s.sensor_type == "structural_health"]
        
        for hour in range(1, min(prediction_horizon, 72) + 1):
            prediction_time = current_time + timedelta(hours=hour)
            
            # Calculate failure probability based on sensor trends
            failure_probability = 0.0
            
            if structural_sensors:
                avg_health = np.mean([s.value for s in structural_sensors[-20:]])  # Last 20 readings
                if avg_health < 0.9:
                    failure_probability = (0.9 - avg_health) * 0.1
            
            # Add random noise for realism
            failure_probability += np.random.normal(0, 0.01)
            failure_probability = max(0, min(1, failure_probability))
            
            confidence = 0.91 * (1 - 0.3 * hour / 72)
            
            prediction = PredictionResult(
                prediction_type="equipment_failure",
                target_time=prediction_time,
                predicted_value=failure_probability,
                confidence=confidence,
                model_version="XGB_v1.5"
            )
            predictions.append(prediction)
        
        return predictions
    
    def get_model_performance_summary(self) -> Dict:
        """Get summary of model performance"""
        return {
            "total_predictions": len(self.prediction_cache),
            "model_accuracy": {
                name: model["accuracy"] 
                for name, model in self.models.items()
            },
            "last_training_updates": {
                name: model["last_updated"] 
                for name, model in self.models.items()
            },
            "prediction_horizons": {
                name: model["horizon_hours"] if "hours" in model else model["horizon_days"]
                for name, model in self.models.items()
            }
        }
```

```python
# Main Digital Twin Simulation
class DigitalTwinSimulation:
    """Main digital twin simulation orchestrator"""
    
    def __init__(self):
        # Initialize the five-layer architecture
        self.architecture = DigitalTwinArchitecture(quay_length=3200, num_berths=8)
        
        # Initialize system integration components
        self.vessel_traffic = VesselTrafficManagementSystem()
        self.cargo_handling = CargoHandlingEquipmentSystem()
        self.yard_operations = YardOperationsSystem()
        self.environmental = EnvironmentalMonitoringSystem()
        
        # Initialize predictive analytics
        self.analytics = PredictiveAnalyticsEngine()
        
        # Simulation parameters
        self.simulation_speed = 1.0  # 1 second = 1 minute real time
        self.data_update_interval = 30  # seconds (from source)
        self.current_time = datetime.now()
        
        # Performance tracking
        self.performance_metrics = {
            "data_points_processed": 0,
            "optimization_decisions": 0,
            "prediction_accuracy": 0.0,
            "system_uptime": 0.0,
            "response_time": 0.0
        }
    
    def run_simulation(self, duration_hours: int) -> Dict:
        """Run the digital twin simulation for specified duration"""
        print(f"=== Digital Twin Simulation: {duration_hours} hours ===")
        print(f"Processing {50000 * duration_hours}+ data points per minute")
        print(f"Integration with 12 port management software systems")
        print()
        
        simulation_results = {
            "timeline": [],
            "kpi_evolution": [],
            "optimization_decisions": [],
            "predictions": [],
            "alerts": [],
            "performance_metrics": {}
        }
        
        start_time = time.time()
        end_time = self.current_time + timedelta(hours=duration_hours)
        
        # Generate initial vessel traffic
        vessels = self._generate_vessel_traffic(duration_hours)
        
        # Simulation loop
        step = 0
        while self.current_time < end_time:
            step_start = time.time()
            
            # Update sensor data (every 30 seconds simulated)
            if step % 30 == 0:
                self._update_sensor_data()
                self.performance_metrics["data_points_processed"] += 50000
            
            # Process vessel arrivals and departures
            self._process_vessel_events(vessels)
            
            # Run optimization (every 5 minutes simulated)
            if step % 300 == 0:
                optimization_result = self._run_optimization()
                simulation_results["optimization_decisions"].append({
                    "timestamp": self.current_time,
                    "result": optimization_result
                })
                self.performance_metrics["optimization_decisions"] += 1
            
            # Update predictions (every hour simulated)
            if step % 3600 == 0:
                predictions = self._update_predictions()
                simulation_results["predictions"].append({
                    "timestamp": self.current_time,
                    "predictions": predictions
                })
            
            # Update KPIs
            self._update_kpis()
            
            # Check for alerts
            alerts = self.architecture.kpi_monitor._generate_alerts()
            if alerts:
                simulation_results["alerts"].append({
                    "timestamp": self.current_time,
                    "alerts": alerts
                })
            
            # Record timeline event
            if step % 1800 == 0:  # Every 30 minutes simulated
                simulation_results["timeline"].append({
                    "timestamp": self.current_time,
                    "berth_utilization": self.architecture.kpi_monitor.kpis["berth_utilization"],
                    "vessel_waiting_time": self.architecture.kpi_monitor.kpis["vessel_waiting_time"],
                    "active_vessels": len([v for v in vessels if v.current_status != VesselStatus.DEPARTED])
                })
            
            # Advance simulation time
            self.current_time += timedelta(seconds=1)
            step += 1
            
            # Control simulation speed
            elapsed = time.time() - step_start
            if elapsed < 0.01:  # Minimum 10ms per step
                time.sleep(0.01 - elapsed)
        
        # Calculate final performance metrics
        total_simulation_time = time.time() - start_time
        self.performance_metrics["system_uptime"] = 100.0  # Assuming no downtime
        self.performance_metrics["response_time"] = total_simulation_time / (duration_hours * 3600)
        
        # Generate final summary
        final_summary = self._generate_simulation_summary(simulation_results)
        simulation_results["performance_metrics"] = self.performance_metrics
        
        return simulation_results
    
    def _generate_vessel_traffic(self, duration_hours: int) -> List[Vessel]:
        """Generate realistic vessel traffic for simulation"""
        vessels = []
        vessel_id = 1
        
        for hour in range(duration_hours):
            # Generate arrivals based on Poisson process
            num_arrivals = np.random.poisson(2.5)  # Average 2.5 arrivals per hour
            
            for _ in range(num_arrivals):
                # Random vessel characteristics
                vessel_types = [
                    ("Feeder", 180, 10, 800),
                    ("Panamax", 250, 12, 2000),
                    ("Post-Panamax", 320, 14, 4000),
                    ("ULCV", 400, 16, 8000)
                ]
                
                vtype = random.choice(vessel_types)
                
                vessel = Vessel(
                    id=f"V_{vessel_id:04d}",
                    name=f"Vessel {vessel_id}",
                    length=vtype[1],
                    draft=vtype[2],
                    cargo_capacity=vtype[3],
                    current_status=VesselStatus.APPROACHING,
                    eta=self.current_time + timedelta(hours=random.uniform(2, 8)),
                    current_position=(random.uniform(40.5, 41.0), random.uniform(-71.0, -70.5)),
                    cargo_type=vtype[0]
                )
                vessels.append(vessel)
                vessel_id += 1
        
        return vessels
    
    def _update_sensor_data(self):
        """Update sensor data with realistic values"""
        current_time = self.current_time
        
        # Update structural health sensors
        for sensor in self.architecture.sensors["structural_health"]:
            sensor.timestamp = current_time
            # Small random variations
            sensor.value = max(0.8, min(1.0, sensor.value + np.random.normal(0, 0.01)))
        
        # Update weather sensors
        for sensor in self.architecture.sensors["weather"]:
            sensor.timestamp = current_time
            sensor.value = max(0, random.uniform(5, 30))  # Wind speed variations
        
        # Update water level sensors
        for sensor in self.architecture.sensors["water_level"]:
            sensor.timestamp = current_time
            # Tidal variations
            tidal_factor = 0.5 * np.sin(2 * np.pi * current_time.hour / 12.42)  # 12.42 hour tidal cycle
            sensor.value = 13.0 + tidal_factor + np.random.normal(0, 0.1)
    
    def _process_vessel_events(self, vessels: List[Vessel]):
        """Process vessel arrivals, departures, and status changes"""
        for vessel in vessels:
            # Check if vessel should arrive
            if vessel.current_status == VesselStatus.APPROACHING and vessel.eta <= self.current_time:
                vessel.current_status = VesselStatus.AT_ANCHOR
            
            # Check if vessel should be berthed
            if vessel.current_status == VesselStatus.AT_ANCHOR:
                # Try to assign berth
                suitable_berths = [
                    b for b in self.architecture.berths
                    if (b.length >= vessel.length and 
                        b.depth >= vessel.draft and 
                        b.current_vessel is None)
                ]
                
                if suitable_berths:
                    best_berth = min(suitable_berths, key=lambda b: b.utilization_rate)
                    vessel.assigned_berth = best_berth.id
                    best_berth.current_vessel = vessel
                    vessel.current_status = VesselStatus.BERTHED
                    
                    # Schedule departure (8-24 hours later)
                    departure_time = self.current_time + timedelta(hours=random.uniform(8, 24))
                    vessel.eta = departure_time  # Reuse eta as departure time
            
            # Check if vessel should depart
            if vessel.current_status == VesselStatus.BERTHED and vessel.eta <= self.current_time:
                if vessel.assigned_berth:
                    berth = next(b for b in self.architecture.berths if b.id == vessel.assigned_berth)
                    berth.current_vessel = None
                    berth.utilization_rate = max(0, berth.utilization_rate - 0.1)
                vessel.current_status = VesselStatus.DEPARTED
    
    def _run_optimization(self) -> Dict:
        """Run real-time optimization for berth allocation"""
        start_time = time.time()
        
        # Get current vessels needing berths
        waiting_vessels = [v for v in self.architecture.vessels 
                         if v.current_status == VesselStatus.AT_ANCHOR]
        
        # Run optimization
        result = self.architecture.optimization_engine.optimize_berth_allocation(
            waiting_vessels, self.architecture.berths, {}
        )
        
        decision_time = time.time() - start_time
        
        return {
            "decision_time": decision_time,
            "vessels_processed": len(waiting_vessels),
            "vessels_assigned": result["vessels_assigned"],
            "optimization_score": result["optimization_score"],
            "within_time_limit": decision_time < 45  # 45 seconds from source
        }
    
    def _update_predictions(self):
        """Update predictive analytics"""
        # Get historical data
        historical_data = []  # Would contain actual historical data
        
        # Update predictions
        arrival_predictions = self.analytics.predict_vessel_arrivals(historical_data, 48)
        cargo_predictions = self.analytics.predict_cargo_volumes(historical_data, 7)
        failure_predictions = self.analytics.predict_equipment_failures(
            self.architecture.sensors["structural_health"], 72
        )
        
        return {
            "arrival_predictions": len(arrival_predictions),
            "cargo_predictions": len(cargo_predictions),
            "failure_predictions": len(failure_predictions),
            "avg_confidence": np.mean([
                np.mean([p.confidence for p in arrival_predictions]),
                np.mean([p.confidence for p in cargo_predictions]),
                np.mean([p.confidence for p in failure_predictions])
            ])
        }
    
    def _update_kpis(self):
        """Update KPI monitor with current data"""
        berth_data = {"berths": self.architecture.berths}
        vessel_data = {"vessels": self.architecture.vessels}
        cargo_data = {}  # Would contain actual cargo data
        
        self.architecture.kpi_monitor.update_kpis(
            berth_data, vessel_data, cargo_data, self.current_time
        )
    
    def _generate_simulation_summary(self, results: Dict) -> Dict:
        """Generate comprehensive simulation summary"""
        
        # Calculate key metrics from results
        if results["timeline"]:
            avg_utilization = np.mean([t["berth_utilization"] for t in results["timeline"]])
            avg_waiting_time = np.mean([t["vessel_waiting_time"] for t in results["timeline"]])
        else:
            avg_utilization = 0.0
            avg_waiting_time = 0.0
        
        # Calculate optimization performance
        if results["optimization_decisions"]:
            avg_decision_time = np.mean([
                d["result"]["decision_time"] 
                for d in results["optimization_decisions"]
            ])
            total_assignments = sum(d["result"]["vessels_assigned"] 
                                 for d in results["optimization_decisions"])
        else:
            avg_decision_time = 0.0
            total_assignments = 0
        
        return {
            "simulation_duration_hours": len(results["timeline"]) / 2,  # Assuming 30min intervals
            "average_berth_utilization": avg_utilization,
            "average_waiting_time_hours": avg_waiting_time,
            "total_optimization_decisions": len(results["optimization_decisions"]),
            "average_decision_time_seconds": avg_decision_time,
            "total_vessels_assigned": total_assignments,
            "total_alerts_generated": len(results["alerts"]),
            "system_performance": self.performance_metrics,
            "roi_improvement": 0.18,  # 18% improvement from source
            "utilization_increase": 0.23,  # 23% increase from source
            "waiting_time_decrease": 0.41   # 41% decrease from source
        }
```

```python
# What-If Analysis and Scenario Evaluation
def what_if_analysis(digital_twin: DigitalTwinSimulation) -> Dict:
    """Perform what-if analysis for infrastructure modification scenarios"""
    
    print("\n" + "="*80)
    print("WHAT-IF ANALYSIS: INFRASTRUCTURE MODIFICATION SCENARIOS")
    print("="*80)
    
    # Define scenarios based on source example
    scenarios = {
        "Baseline": {
            "description": "Current configuration (8 berths, 3.2km)",
            "modifications": {},
            "expected_impact": "Reference case"
        },
        "Berth_6_Extension": {
            "description": "Extend Berth 6 by 85 meters",
            "modifications": {"berth_6_length": 495},  # 410 + 85
            "expected_impact": "NPV return: $12.4 million over 15 years"
        },
        "Additional_Berth": {
            "description": "Add 9th berth (400m)",
            "modifications": {"num_berths": 9, "total_length": 3600},
            "expected_impact": "Increased capacity, reduced waiting times"
        },
        "Deepening_Project": {
            "description": "Deepen all berths by 2 meters",
            "modifications": {"depth_increase": 2.0},
            "expected_impact": "Accommodate larger vessels"
        },
        "Combined_Upgrade": {
            "description": "Berth 6 extension + 9th berth + deepening",
            "modifications": {
                "berth_6_length": 495,
                "num_berths": 9,
                "total_length": 3600,
                "depth_increase": 2.0
            },
            "expected_impact": "Comprehensive capacity enhancement"
        }
    }
    
    scenario_results = {}
    
    for scenario_name, scenario_config in scenarios.items():
        print(f"\n📋 {scenario_name}:")
        print(f"   {scenario_config['description']}")
        print(f"   Expected: {scenario_config['expected_impact']}")
        
        # Simulate scenario (simplified for demonstration)
        result = simulate_scenario(digital_twin, scenario_config)
        scenario_results[scenario_name] = result
        
        print(f"   ✅ Berth Utilization: {result['berth_utilization']*100:.1f}%")
        print(f"   ✅ Waiting Time: {result['waiting_time']:.1f} hours")
        print(f"   ✅ Annual Revenue: ${result['annual_revenue']:,.0f}")
        print(f"   ✅ NPV (15 years): ${result['npv']:,.0f}")
        print(f"   ✅ Payback Period: {result['payback_period']:.1f} years")
    
    # Create scenario comparison visualization
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
    
    scenario_names = list(scenario_results.keys())
    
    # Plot 1: Berth utilization comparison
    utilizations = [scenario_results[name]['berth_utilization'] * 100 for name in scenario_names]
    bars = ax1.bar(scenario_names, utilizations, color=['#2E86AB', '#A23B72', '#F18F01', '#C73E1D', '#E71D36'], alpha=0.8)
    ax1.set_ylabel('Berth Utilization (%)')
    ax1.set_title('Berth Utilization by Scenario')
    ax1.grid(True, alpha=0.3, axis='y')
    
    for bar, util in zip(bars, utilizations):
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2., height + 1,
                f'{util:.1f}%', ha='center', va='bottom', fontweight='bold')
    
    # Plot 2: Waiting time comparison
    waiting_times = [scenario_results[name]['waiting_time'] for name in scenario_names]
    bars = ax2.bar(scenario_names, waiting_times, color=['#2E86AB', '#A23B72', '#F18F01', '#C73E1D', '#E71D36'], alpha=0.8)
    ax2.set_ylabel('Average Waiting Time (hours)')
    ax2.set_title('Vessel Waiting Time by Scenario')
    ax2.grid(True, alpha=0.3, axis='y')
    
    for bar, wait in zip(bars, waiting_times):
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                f'{wait:.1f}h', ha='center', va='bottom', fontweight='bold')
    
    # Plot 3: NPV comparison
    npvs = [scenario_results[name]['npv'] / 1e6 for name in scenario_names]  # Convert to millions
    bars = ax3.bar(scenario_names, npvs, color=['#2E86AB', '#A23B72', '#F18F01', '#C73E1D', '#E71D36'], alpha=0.8)
    ax3.set_ylabel('NPV ($ millions)')
    ax3.set_title('15-Year NPV by Scenario')
    ax3.grid(True, alpha=0.3, axis='y')
    
    for bar, npv in zip(bars, npvs):
        height = bar.get_height()
        ax3.text(bar.get_x() + bar.get_width()/2., height + 0.5,
                f'${npv:.1f}M', ha='center', va='bottom', fontweight='bold')
    
    # Plot 4: Payback period comparison
    payback_periods = [scenario_results[name]['payback_period'] for name in scenario_names]
    bars = ax4.bar(scenario_names, payback_periods, color=['#2E86AB', '#A23B72', '#F18F01', '#C73E1D', '#E71D36'], alpha=0.8)
    ax4.set_ylabel('Payback Period (years)')
    ax4.set_title('Payback Period by Scenario')
    ax4.grid(True, alpha=0.3, axis='y')
    
    for bar, period in zip(bars, payback_periods):
        height = bar.get_height()
        ax4.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                f'{period:.1f}y', ha='center', va='bottom', fontweight='bold')
    
    plt.suptitle('What-If Analysis: Infrastructure Modification Scenarios', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # Find best scenario
    best_scenario = max(scenario_results.keys(), key=lambda k: scenario_results[k]['npv'])
    print(f"\n🏆 RECOMMENDED SCENARIO: {best_scenario}")
    print(f"   NPV: ${scenario_results[best_scenario]['npv']:,.0f}")
    print(f"   Payback: {scenario_results[best_scenario]['payback_period']:.1f} years")
    
    return scenario_results

def simulate_scenario(digital_twin: DigitalTwinSimulation, scenario_config: Dict) -> Dict:
    """Simulate a specific what-if scenario"""
    
    # Base metrics from current configuration
    base_utilization = digital_twin.architecture.kpi_monitor.kpis.get("berth_utilization", 0.75)
    base_waiting_time = digital_twin.architecture.kpi_monitor.kpis.get("vessel_waiting_time", 4.0)
    base_annual_revenue = 50000000  # $50M base annual revenue
    
    # Apply scenario modifications
    modifications = scenario_config.get("modifications", {})
    
    # Calculate impacts
    utilization_impact = 0.0
    waiting_time_impact = 0.0
    revenue_impact = 0.0
    investment_cost = 0.0
    
    if "berth_6_length" in modifications:
        # Berth 6 extension: +5% capacity, -10% waiting time, +$2M revenue
        utilization_impact += 0.05
        waiting_time_impact -= 0.10
        revenue_impact += 2000000
        investment_cost += 8500000  # $8.5M for 85m extension
    
    if "num_berths" in modifications:
        # Additional berth: +15% capacity, -25% waiting time, +$5M revenue
        utilization_impact += 0.15
        waiting_time_impact -= 0.25
        revenue_impact += 5000000
        investment_cost += 15000000  # $15M for new berth
    
    if "depth_increase" in modifications:
        # Deepening: +8% capacity (larger vessels), -5% waiting time, +$3M revenue
        utilization_impact += 0.08
        waiting_time_impact -= 0.05
        revenue_impact += 3000000
        investment_cost += 8000000  # $8M for deepening
    
    # Calculate new metrics
    new_utilization = min(0.95, base_utilization + utilization_impact)
    new_waiting_time = max(0.5, base_waiting_time + waiting_time_impact)
    new_annual_revenue = base_annual_revenue + revenue_impact
    
    # Calculate NPV (15 years, 8% discount rate)
    discount_rate = 0.08
    years = 15
    
    npv = -investment_cost
    for year in range(1, years + 1):
        npv += new_annual_revenue / ((1 + discount_rate) ** year)
    
    # Calculate payback period
    cumulative_cash_flow = -investment_cost
    payback_period = years  # Default if never pays back
    
    for year in range(1, years + 1):
        cumulative_cash_flow += new_annual_revenue
        if cumulative_cash_flow >= 0:
            payback_period = year - (cumulative_cash_flow - new_annual_revenue) / new_annual_revenue
            break
    
    return {
        "berth_utilization": new_utilization,
        "waiting_time": new_waiting_time,
        "annual_revenue": new_annual_revenue,
        "npv": npv,
        "payback_period": payback_period,
        "investment_cost": investment_cost
    }

# Run what-if analysis
what_if_results = what_if_analysis(digital_twin)
```
