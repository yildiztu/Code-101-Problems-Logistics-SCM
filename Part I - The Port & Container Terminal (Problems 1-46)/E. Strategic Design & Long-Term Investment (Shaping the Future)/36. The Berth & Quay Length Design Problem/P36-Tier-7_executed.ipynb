## Tier 7: The Human-AI Symbiotic Partnership (The Centaur Model)

### Goal
Implement a human-AI symbiotic partnership framework that leverages the complementary strengths of human expertise and artificial intelligence for optimal berth design decisions, ensuring ethical considerations and stakeholder alignment.

### Key assumptions
- Human experts provide strategic vision, stakeholder management, and creative problem-solving
- AI systems provide computational optimization, pattern recognition, and rapid scenario analysis
- Explainable AI (XAI) provides transparent decision reasoning to build trust
- Confidence-based human intervention protocols allow humans to override AI when needed
- Ethical principles are embedded directly into algorithmic decision-making processes
- Learning loop from human-AI interactions continuously improves system performance

### Approach (step-by-step)
1. Design the centaur model architecture with complementary intelligence layers
2. Implement explainable AI systems with decision tree visualization and feature importance analysis
3. Create interactive optimization dashboard with constraint negotiation interface
4. Develop collaborative what-if analysis tools for joint human-AI exploration
5. Implement ethical constraint integration with environmental justice and precautionary principles
6. Create trust metrics monitoring and performance validation systems
7. Demonstrate the framework with real-world port expansion scenarios

- Ethical constraint satisfaction and stakeholder approval ratings

```python
# Define data structures for Human-AI Symbiotic Partnership
from dataclasses import dataclass
from typing import List, Dict, Tuple, Optional, Any
from enum import Enum

class EthicalPrinciple(Enum):
    """Ethical principles for AI decision-making"""
    NON_MALEFICENCE = "non_maleficence"  # Prevent harm to crew, environment, cargo
    BENEFICENCE = "beneficence"          # Promote safety excellence
    AUTONOMY = "autonomy"              # Respect stakeholder decision authority
    JUSTICE = "justice"                # Fair treatment across stakeholders
    TRANSPARENCY = "transparency"      # Explainable reasoning
    PRECAUTIONARY = "precautionary"    # Apply stricter standards when uncertain

@dataclass
class Stakeholder:
    """Represents a human stakeholder in the decision process"""
    id: str
    name: str
    role: str  # Port Development Director, Environmental Analyst, etc.
    expertise: List[str]  # Areas of expertise
    priorities: Dict[str, float]  # Priority weights for different objectives
    confidence_threshold: float  # Minimum confidence to accept AI recommendations

@dataclass
class BerthConfiguration:
    """Represents a berth configuration solution"""
    id: str
    num_berths: int
    berth_lengths: List[float]  # meters
    total_length: float  # meters
    construction_cost: float  # dollars
    operational_cost: float  # dollars per year
    waiting_cost: float  # dollars per year
    total_cost: float  # dollars
    
    # AI system evaluation
    ai_optimization_score: float  # 0-1 scale
    ai_confidence: float  # 0-1 scale
    
    # Human expert evaluation
    human_feasibility_score: float  # 0-1 scale
    human_stakeholder_approval: float  # 0-1 scale
    human_ethical_compliance: float  # 0-1 scale
    
    # Ethical constraint scores
    environmental_justice: float  # 0-1 scale
    worker_safety_equity: float  # 0-1 scale
    transparency_score: float  # 0-1 scale
    precautionary_compliance: float  # 0-1 scale
```

```python
try:
    # Explainable AI System Implementation
    class ExplainableAISystem:
        """Explainable AI system for transparent berth design decisions"""
        
        def __init__(self):
            self.ml_models = {
                "multi_objective_optimizer": {
                    "model_type": "Pareto_Front_Explorer",
                    "objectives": ["cost_minimization", "capacity_maximization", "waiting_time_reduction", "flexibility_maximization"],
                    "last_updated": datetime.now() - timedelta(days=7)
                },
                "scenario_analyzer": {
                    "model_type": "Monte_Carlo_Simulation",
                    "scenarios_per_hour": 10000,
                    "confidence_calculation": "statistical_bounds"
                },
                "ethical_compliance_checker": {
                    "model_type": "Rule_Based_Validator",
                    "ethical_principles": [p.value for p in EthicalPrinciple],
                    "compliance_threshold": 0.8
                }
            }
        
        def generate_recommendation(self, problem_constraints: Dict, 
                                    stakeholder_priorities: Dict[str, float]) -> Dict:
            """Generate AI recommendation with explainable reasoning"""
            
            start_time = time.time()
            
            # Step 1: Generate multiple berth configuration alternatives
            alternatives = self._generate_alternatives(problem_constraints)
            
            # Step 2: Evaluate each alternative across multiple objectives
            evaluated_alternatives = []
            for alt in alternatives:
                evaluation = self._evaluate_alternative(alt, stakeholder_priorities)
                evaluated_alternatives.append(evaluation)
            
            # Step 3: Select best alternative using Pareto optimality
            best_alternative = self._select_pareto_optimal(evaluated_alternatives)
            
            # Step 4: Generate explainable reasoning
            reasoning = self._generate_reasoning(best_alternative, evaluated_alternatives)
            
            # Step 5: Calculate confidence intervals
            confidence = self._calculate_confidence(best_alternative, evaluated_alternatives)
            
            computation_time = time.time() - start_time
        
            return {
                "configuration": best_alternative,
                "reasoning": reasoning,
                "confidence": confidence,
                "alternatives_considered": len(alternatives),
                "computation_time": computation_time
            }
        
        def _generate_alternatives(self, constraints: Dict) -> List[BerthConfiguration]:
            """Generate multiple berth configuration alternatives"""
            alternatives = []
            
            # Generate diverse configurations
            num_berths_options = [8, 10, 12, 15, 18, 20]
            
            for num_berths in num_berths_options:
                if num_berths * constraints["max_berth_length"] > constraints["max_total_length"]:
                    continue
                
                # Generate different length distribution strategies
                strategies = ["equal", "weighted", "mixed", "progressive"]
                
                for strategy in strategies:
                    if strategy == "equal":
                        berth_length = constraints["max_total_length"] / num_berths
                        berth_lengths = [berth_length] * num_berths
                    elif strategy == "weighted":
                        # Weight towards larger berths for bigger vessels
                        weights = np.linspace(0.8, 1.2, num_berths)
                        total_weight = sum(weights)
                        berth_lengths = [(constraints["max_total_length"] * w / total_weight) for w in weights]
                    elif strategy == "mixed":
                        # Mix of small, medium, and large berths
                        small_berths = max(2, num_berths // 3)
                        large_berths = max(1, num_berths // 4)
                        medium_berths = num_berths - small_berths - large_berths
                        
                        lengths = []
                        lengths.extend([280] * small_berths)  # Small berths
                        lengths.extend([350] * medium_berths)  # Medium berths
                        lengths.extend([420] * large_berths)  # Large berths
                        
                        # Adjust to meet total length constraint
                        scale_factor = constraints["max_total_length"] / sum(lengths)
                        berth_lengths = [l * scale_factor for l in lengths]
                    else:  # progressive
                        # Gradually increasing sizes
                        base_length = constraints["min_berth_length"]
                        berth_lengths = []
                        for i in range(num_berths):
                            length = base_length + (i * 15)  # 15m increment per berth
                            berth_lengths.append(length)
                        
                        # Adjust to meet total length constraint
                        scale_factor = constraints["max_total_length"] / sum(berth_lengths)
                        berth_lengths = [l * scale_factor for l in berth_lengths]
                    
                    # Ensure constraints are met
                    if (sum(berth_lengths) <= constraints["max_total_length"] and
                        all(l >= constraints["min_berth_length"] and l <= constraints["max_berth_length"] for l in berth_lengths)):
                        
                        # Calculate costs
                        construction_cost = sum(berth_lengths) * 50000  # 50,000 euro per meter
                        operational_cost = self._estimate_operational_cost(berth_lengths)
                        waiting_cost = self._estimate_waiting_cost(berth_lengths)
                        total_cost = construction_cost + operational_cost + waiting_cost
                        
                        config = BerthConfiguration(
                            id=f"AI_{len(alternatives)+1:03d}",
                            num_berths=num_berths,
                            berth_lengths=berth_lengths,
                            total_length=sum(berth_lengths),
                            construction_cost=construction_cost,
                            operational_cost=operational_cost,
                            waiting_cost=waiting_cost,
                            total_cost=total_cost,
                            ai_optimization_score=0.0,
                            ai_confidence=0.0,
                            human_feasibility_score=0.0,
                            human_stakeholder_approval=0.0,
                            human_ethical_compliance=0.0,
                            environmental_justice=0.0,
                            worker_safety_equity=0.0,
                            transparency_score=0.0,
                            precautionary_compliance=0.0
                        )
                        
                        alternatives.append(config)
            
            return alternatives
        
        def _evaluate_alternative(self, config: BerthConfiguration, 
                               priorities: Dict[str, float]) -> Dict:
            """Evaluate a berth configuration across multiple objectives"""
            
            # Normalize scores
            # Cost objective (lower is better)
            cost_score = 1.0 - (config.total_cost / 100000000)  # Normalize to 100M euro max
            cost_score = max(0, cost_score)
            
            # Capacity objective (higher is better)
            capacity_score = config.num_berths / 20.0  # Normalize to 20 berths max
            
            # Flexibility objective (higher is better)
            # Calculate flexibility based on berth length diversity
            length_variance = np.var(config.berth_lengths) if len(config.berth_lengths) > 1 else 0
            max_variance = ((constraints["max_berth_length"] - constraints["min_berth_length"]) ** 2) / 4
            flexibility_score = length_variance / max_variance
            
            # Weighted combination based on stakeholder priorities
            total_score = 0.0
            
            # Map stakeholder priorities to objectives
            priority_mapping = {
                "economic_efficiency": {"cost": 0.7, "capacity": 0.2, "flexibility": 0.1},
                "capacity_growth": {"cost": 0.1, "capacity": 0.7, "flexibility": 0.2},
                "risk_management": {"cost": 0.3, "capacity": 0.3, "flexibility": 0.4},
                "innovation": {"cost": 0.2, "capacity": 0.3, "flexibility": 0.5},
                "stakeholder_relations": {"cost": 0.4, "capacity": 0.3, "flexibility": 0.3},
                "safety": {"cost": 0.1, "capacity": 0.4, "flexibility": 0.3, "waiting_time": 0.2},
                "technical_feasibility": {"cost": 0.2, "capacity": 0.3, "flexibility": 0.3, "safety": 0.2},
                "environmental_protection": {"cost": 0.1, "capacity": 0.2, "flexibility": 0.4, "environmental_justice": 0.3},
                "regulatory_compliance": {"cost": 0.1, "capacity": 0.2, "flexibility": 0.3, "environmental_justice": 0.4},
                "biodiversity": {"cost": 0.1, "capacity": 0.1, "flexibility": 0.4, "environmental_justice": 0.4},
                "operational_efficiency": {"cost": 0.2, "capacity": 0.3, "flexibility": 0.2, "waiting_time": 0.3},
                "throughput_maximization": {"cost": 0.1, "capacity": 0.4, "flexibility": 0.3, "waiting_time": 0.2},
                "equipment_utilization": {"cost": 0.1, "capacity": 0.3, "flexibility": 0.3, "waiting_time": 0.3},
                "cost_minimization": {"cost": 0.8, "capacity": 0.1, "flexibility": 0.1},
                "roi_maximization": {"cost": 0.3, "capacity": 0.4, "flexibility": 0.2, "long_term_value": 0.1},
                "budget_compliance": {"cost": 0.6, "capacity": 0.2, "flexibility": 0.1, "long_term_value": 0.1},
                "long_term_value": {"cost": 0.2, "capacity": 0.3, "flexibility": 0.2, "innovation": 0.3}
            }
            
            # Calculate weighted score for each stakeholder
            stakeholder_scores = []
            for stakeholder_priority in priorities.values():
                if stakeholder_priority in priority_mapping:
                    weights = priority_mapping[stakeholder_priority]
                    score = (weights["cost"] * cost_score + 
                           weights["capacity"] * capacity_score + 
                           weights["flexibility"] * flexibility_score)
                    stakeholder_scores.append(score)
            
            # Average stakeholder score
            if stakeholder_scores:
                total_score = np.mean(stakeholder_scores)
            else:
                total_score = 0.5  # Default if no priorities provided
            
            return {
                "configuration": config,
                "cost_score": cost_score,
                "capacity_score": capacity_score,
                "flexibility_score": flexibility_score,
                "total_score": total_score,
                "stakeholder_scores": stakeholder_scores
            }
        
        def _select_pareto_optimal(self, evaluated: List[Dict]) -> BerthConfiguration:
            """Select Pareto optimal solution from evaluated alternatives"""
            
            # Extract configurations and scores
            configs = [e["configuration"] for e in evaluated]
            cost_scores = [1 - e["cost_score"] for e in evaluated]  # Convert back to original scale
            capacity_scores = [e["capacity_score"] for e in evaluated]
            flexibility_scores = [e["flexibility_score"] for e in evaluated]
            total_scores = [e["total_score"] for e in evaluated]
            
            # Find Pareto frontier (simplified approach)
            pareto_indices = []
            
            for i, config in enumerate(configs):
                is_pareto = True
                
                # Check if any other configuration is strictly better
                for j, other_config in enumerate(configs):
                    if i != j:
                        if (cost_scores[j] <= cost_scores[i] and 
                            capacity_scores[j] >= capacity_scores[i] and 
                            flexibility_scores[j] >= flexibility_scores[i] and
                            (cost_scores[j] < cost_scores[i] or 
                             capacity_scores[j] > capacity_scores[i] or 
                             flexibility_scores[j] > flexibility_scores[i])):
                            is_pareto = False
                            break
                
                if is_pareto:
                    pareto_indices.append(i)
            
            # Select best Pareto solution (balanced approach)
            if pareto_indices:
                # Use weighted sum of objectives to select from Pareto frontier
                best_idx = max(pareto_indices, 
                               key=lambda i: (0.4 * cost_scores[i] + 
                                            0.4 * capacity_scores[i] + 
                                            0.2 * flexibility_scores[i]))
                return configs[best_idx]
            else:
                # Fallback to best total score
                best_idx = max(range(len(configs)), key=lambda i: total_scores[i])
                return configs[best_idx]
        
        def _generate_reasoning(self, best_config: BerthConfiguration, 
                               all_evaluated: List[Dict]) -> List[str]:
            """Generate explainable reasoning for the recommendation"""
            
            reasoning = []
            
            reasoning.append(f"Selected configuration {best_config.id} based on Pareto optimality analysis.")
            reasoning.append(f"Configuration: {best_config.num_berths} berths with lengths {[f'{l:.0f}m' for l in best_config.berth_lengths]}")
            reasoning.append(f"Total Length: {best_config.total_length:.0f} meters (constraint: {constraints['max_total_length']}m)")
            
            # Compare with alternatives
            if len(all_evaluated) > 1:
                reasoning.append("\nComparison with evaluated alternatives:")
                
                # Show cost comparison
                cost_comparison = sorted(all_evaluated, key=lambda x: x["configuration"].total_cost)
                reasoning.append(f"  Cost ranking: {cost_comparison[0]['configuration'].id} (€{cost_comparison[0]['configuration'].total_cost/1e6:.1f}M) to {cost_comparison[-1]['configuration'].id} (€{cost_comparison[-1]['configuration'].total_cost/1e6:.1f}M)")
                
                # Show capacity comparison
                capacity_comparison = sorted(all_evaluated, key=lambda x: x["configuration"].num_berths, reverse=True)
                reasoning.append(f"  Capacity ranking: {capacity_comparison[0]['configuration'].id} ({capacity_comparison[0]['configuration'].num_berths} berths) to {capacity_comparison[-1]['configuration'].id} ({capacity_comparison[-1]['configuration'].num_berths} berths)")
            
            # Explain trade-offs
            reasoning.append("\nKey trade-offs considered:")
            reasoning.append(f"  • Economic efficiency: €{best_config.total_cost/1e6:.1f}M total investment")
            reasoning.append(f"  • Capacity provision: {best_config.num_berths} berths for vessel accommodation")
            reasoning.append(f"  • Operational flexibility: Diverse berth lengths for different vessel types")
            reasoning.append(f"  • Constraint satisfaction: All physical and operational constraints met")
            
            return reasoning
        
        def _calculate_confidence(self, best_config: BerthConfiguration, 
                                 all_evaluated: List[Dict]) -> float:
            """Calculate confidence intervals for the recommendation"""
            
            # Base confidence from model performance
            base_confidence = 0.85  # 85% base confidence from ML models
            
            # Adjust confidence based on consensus among alternatives
            if len(all_evaluated) > 1:
                # Calculate variance in objective scores
                score_variance = np.var([e["total_score"] for e in all_evaluated])
                max_variance = 0.25  # Maximum expected variance
                
                # Reduce confidence if high variance (disagreement among alternatives)
                consensus_penalty = score_variance / max_variance
                confidence_consensus = base_confidence * (1 - consensus_penalty)
            else:
                confidence_consensus = base_confidence
            
            # Adjust confidence based on constraint tightness
            length_utilization = best_config.total_length / constraints["max_total_length"]
            if length_utilization > 0.9:
                length_penalty = (length_utilization - 0.9) * 0.1
            else:
                length_penalty = 0
            
            final_confidence = confidence_consensus * (1 - length_penalty)
            return max(0.5, min(0.95, final_confidence))
        
        def _estimate_operational_cost(self, berth_lengths: List[float]) -> float:
            """Estimate annual operational costs"""
            # Simplified operational cost model
            base_cost_per_meter = 5000  # 5,000 euro per meter per year
            num_berths = len(berth_lengths)
            
            # Base cost
            operational_cost = sum(berth_lengths) * base_cost_per_meter
            
            # Add complexity cost for more berths
            if num_berths > 10:
                complexity_factor = 1.0 + (num_berths - 10) * 0.02
                operational_cost *= complexity_factor
            
            return operational_cost
        
        def _estimate_waiting_cost(self, berth_lengths: List[float]) -> float:
            """Estimate vessel waiting costs"""
            # Simplified waiting cost model
            total_capacity = sum(berth_lengths) / 300  # Assume 1 berth serves ~300m of vessel length
            
            # Base waiting cost per vessel
            base_waiting_cost_per_vessel = 10000  # 10,000 euro per vessel per year
            
            # Adjust for capacity utilization
            utilization_factor = min(1.0, total_capacity / 1000)  # Assume 1000 vessel capacity max
            
            if utilization_factor > 0.8:
                # Exponential cost increase for high utilization
                waiting_multiplier = 1.0 + 5.0 * ((utilization_factor - 0.8) ** 2)
            else:
                waiting_multiplier = 1.0
            
            estimated_annual_vessels = 500  # 500 vessels per year
            
            waiting_cost = estimated_annual_vessels * base_waiting_cost_per_vessel * waiting_multiplier
            
            return waiting_cost
    
    # Create and test the AI system
    ai_system = ExplainableAISystem()
    
    # Generate a recommendation
    recommendation = ai_system.generate_recommendation(
        constraints,
        {"economic_efficiency": 0.3, "capacity_growth": 0.25, "risk_management": 0.2, "innovation": 0.15, "stakeholder_relations": 0.1}
    )
    
    print("AI RECOMMENDATION:")
    print(f"Configuration: {recommendation['configuration'].id}")
    print(f"Berths: {recommendation['configuration'].num_berths}")
    print(f"Total Length: {recommendation['configuration'].total_length:.1f} meters")
    print(f"Total Cost: €{recommendation['configuration'].total_cost/1e6:.2f} million")
    print(f"AI Confidence: {recommendation['confidence']*100:.1f}%")
    print(f"Computation Time: {recommendation['computation_time']:.2f} seconds")
    print(f"Alternatives Considered: {recommendation['alternatives_considered']}")
    print()
    
    print("AI REASONING:")
    for i, reason in enumerate(recommendation['reasoning'], 1):
        print(f"  {i}. {reason}")
except Exception as e:
    print(f'Error in cell: {e}')
```
