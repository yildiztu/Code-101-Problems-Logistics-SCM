# 42. The Peak Energy Consumption & Load Shifting Problem

## Tier 8: The Value-Aligned & Ethical Framework

### Goal
Implement a value-aligned energy management system that transcends pure optimization to incorporate ethical principles, stakeholder values, and long-term sustainability goals, ensuring that energy optimization decisions balance economic efficiency with environmental protection, worker welfare, and community benefits.

### Key assumptions
- Energy optimization must balance multiple stakeholder interests
- Ethical constraints are as important as efficiency metrics
- Long-term sustainability outweighs short-term cost savings
- Environmental justice and equity are non-negotiable requirements
- Stakeholder participation enhances decision legitimacy
- Worker safety cannot be compromised for energy savings
- Community welfare must be protected during optimization

### Approach (step-by-step)
1. Define multi-layered ethical framework with constitutional principles
2. Implement stakeholder value balancing mechanisms
3. Create environmental protection vs economic efficiency trade-offs
4. Develop energy justice and fairness analysis algorithms
5. Build ethical constraint enforcement and violation detection
6. Implement stakeholder feedback integration systems
7. Create value-aligned scoring and decision recommendation
8. Validate system with real-world ethical scenarios

### What to look for in the results
- Ethical constraint compliance and violation detection
- Stakeholder value balancing and trade-off analysis
- Environmental justice impact distribution
- Value-aligned scoring vs pure optimization metrics
- Long-term sustainability vs short-term efficiency analysis
- Ethical decision recommendations with transparency
- Community protection during extreme weather events

### Concrete example (from the source)
Value-aligned optimization during a severe heat wave grid emergency:

**Situation:** Grid emergency requires 15 MW demand reduction. Traditional optimization would curtail community services (hospital backup power, elderly care facilities) while maintaining profitable industrial operations.

**Constitutional AI Response:**
- Healthcare facilities: Absolute protection (constitutional constraint)
- Elderly care centers: Absolute protection (human welfare constraint)
- School cooling systems: High protection (vulnerable population constraint)
- Community centers: Moderate protection (social equity constraint)

**Ethical Optimization Results:**
- Industrial crane operations: 8 MW reduction (rescheduled to night shift)
- Non-essential warehouse lighting: 2.5 MW reduction
- EV charging (non-emergency vehicles): 3 MW reduction (delayed 4 hours)
- Administrative building HVAC: 1.5 MW reduction (temperature setpoint adjustment)
- Community services: 0 MW reduction (fully protected)

**Outcomes:**
- Total cost increase: 12.4% ($4,800) compared to pure economic optimization
- Community welfare maintained: 100% critical services protected
- Stakeholder satisfaction: 8.7/10 vs. 4.2/10 for economic-only approach
- Worker safety score: Maintained at 9.2/10 vs. potential 6.8/10
- Environmental justice index: Improved from baseline due to industrial load shifting

```python
from typing import Tuple, List, Dict, Set
from itertools import product

# Import required libraries for Value-Aligned & Ethical Framework
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass
from typing import List, Tuple, Dict, Optional, Any
import random
from enum import Enum
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('default')
sns.set_palette("husl")
```

```python
class EthicalPrinciple(Enum):
    """Core ethical principles for energy management"""
    ENVIRONMENTAL_PROTECTION = "environmental_protection"
    ENERGY_JUSTICE = "energy_justice"
    WORKER_SAFETY = "worker_safety"
    COMMUNITY_WELFARE = "community_welfare"
    ECONOMIC_SUSTAINABILITY = "economic_sustainability"
    TRANSPARENCY = "transparency"

@dataclass
class StakeholderGroup:
    """Represents a stakeholder group with values and weights"""
    name: str
    weight: float
    values: Dict[str, float]
    impact_sensitivity: float
    vulnerability_score: float  # 0-1, higher = more vulnerable

@dataclass
class EthicalConstraint:
    """Represents an ethical constraint with threshold"""
    principle: EthicalPrinciple
    threshold: float
    is_hard_constraint: bool
    description: str
    penalty_weight: float

@dataclass
class EnergyScenarioImpact:
    """Represents the impact of an energy scenario on different aspects"""
    cost_reduction: float  # percentage
    carbon_emissions: float  # kg CO₂e/day
    community_health: float  # -1 to 1 scale
    worker_safety: float  # -1 to 1 scale
    energy_justice: float  # -1 to 1 scale
    long_term_sustainability: float  # -1 to 1 scale
    grid_stability: float  # -1 to 1 scale

@dataclass
class EthicalEnergyScenario:
    """Represents an ethical energy scenario for evaluation"""
    id: str
    name: str
    description: str
    cost_increase: float  # percentage
    implementation_time: str
    impact: EnergyScenarioImpact
    value_aligned_score: float
    compliance_status: str
    ethical_violations: List[str]
```

```python
try:
    class ValueAlignedEnergyManagement:
        """Value-Aligned Energy Management System with Ethical Framework"""
        
        def __init__(self):
            self.stakeholders = self._initialize_stakeholders()
            self.constraints = self._initialize_constraints()
            self.scenarios = []
            self.carbon_budget = 2850  # tCO₂/year baseline
            self.reduction_target = 0.183  # 18.3% reduction required
        
        def _initialize_stakeholders(self) -> List[StakeholderGroup]:
            """Initialize stakeholder groups with their values and weights"""
            return [
                StakeholderGroup(
                    name="Community",
                    weight=0.30,
                    values={
                        "energy_affordability": 0.9,
                        "air_quality": 0.8,
                        "health_protection": 0.9,
                        "emergency_services": 0.95
                    },
                    impact_sensitivity=0.85,
                    vulnerability_score=0.8
                ),
                StakeholderGroup(
                    name="Workers",
                    weight=0.25,
                    values={
                        "workplace_safety": 0.95,
                        "job_security": 0.8,
                        "working_conditions": 0.7,
                        "training_opportunities": 0.6
                    },
                    impact_sensitivity=0.9,
                    vulnerability_score=0.6
                ),
                StakeholderGroup(
                    name="Environment",
                    weight=0.25,
                    values={
                        "carbon_reduction": 0.9,
                        "renewable_energy": 0.8,
                        "biodiversity": 0.7,
                        "water_conservation": 0.6
                    },
                    impact_sensitivity=0.7,
                    vulnerability_score=0.9
                ),
                StakeholderGroup(
                    name="Operations",
                    weight=0.20,
                    values={
                        "operational_efficiency": 0.8,
                        "service_reliability": 0.9,
                        "cost_effectiveness": 0.7,
                        "productivity": 0.8
                    },
                    impact_sensitivity=0.6,
                    vulnerability_score=0.3
                )
            ]
        
        def _initialize_constraints(self) -> List[EthicalConstraint]:
            """Initialize constitutional and ethical constraints"""
            return [
                # Constitutional constraints (non-negotiable)
                EthicalConstraint(
                    principle=EthicalPrinciple.WORKER_SAFETY,
                    threshold=0.85,  # Minimum 85% safety score
                    is_hard_constraint=True,
                    description="Worker safety cannot be compromised for energy savings",
                    penalty_weight=1.0
                ),
                EthicalConstraint(
                    principle=EthicalPrinciple.COMMUNITY_WELFARE,
                    threshold=0.90,  # Minimum 90% community service protection
                    is_hard_constraint=True,
                    description="Essential community services must be protected",
                    penalty_weight=1.0
                ),
                EthicalConstraint(
                    principle=EthicalPrinciple.ENVIRONMENTAL_PROTECTION,
                    threshold=self.carbon_budget * (1 - self.reduction_target),
                    is_hard_constraint=True,
                    description=f"Carbon emissions must not exceed {self.carbon_budget * (1 - self.reduction_target):.0f} tCO₂/year",
                    penalty_weight=1.0
                ),
                
                # Policy constraints (high priority)
                EthicalConstraint(
                    principle=EthicalPrinciple.ENERGY_JUSTICE,
                    threshold=0.70,  # Minimum 70% energy justice score
                    is_hard_constraint=False,
                    description="Energy allocation must be fair across demographic groups",
                    penalty_weight=0.8
                ),
                EthicalConstraint(
                    principle=EthicalPrinciple.ECONOMIC_SUSTAINABILITY,
                    threshold=0.50,  # Maximum 50% cost increase
                    is_hard_constraint=False,
                    description="Cost increase should not exceed 50% for sustainability",
                    penalty_weight=0.6
                ),
                EthicalConstraint(
                    principle=EthicalPrinciple.GRID_STABILITY,
                    threshold=0.80,  # Minimum 80% grid stability contribution
                    is_hard_constraint=False,
                    description="Should support grid stability and reliability",
                    penalty_weight=0.7
                )
            ]
        
        def create_ethical_scenarios(self) -> List[EthicalEnergyScenario]:
            """Create ethical scenarios for evaluation based on source material"""
            
            scenarios = [
                # Scenario 1: Pure Economic Optimization (violates ethical constraints)
                EthicalEnergyScenario(
                    id="pure_optimization",
                    name="Pure Economic Optimization",
                    description="Maximum cost reduction without ethical considerations",
                    cost_increase=-12.4,  # Actually reduces costs
                    implementation_time="immediate",
                    impact=EnergyScenarioImpact(
                        cost_reduction=25.3,
                        carbon_emissions=2120,  # 18.3% below baseline
                        community_health=-0.40,  # Negative impact on vulnerable communities
                        worker_safety=0.68,  # Below safety threshold
                        energy_justice=0.45,  # Poor energy justice
                        long_term_sustainability=0.60,
                        grid_stability=0.85
                    ),
                    value_aligned_score=0.0,
                    compliance_status="VIOLATION",
                    ethical_violations=[]
                ),
                
                # Scenario 2: Value-Aligned Optimization (from source material)
                EthicalEnergyScenario(
                    id="value_aligned",
                    name="Value-Aligned Optimization",
                    description="Ethical optimization during extreme weather event",
                    cost_increase=12.4,
                    implementation_time="medium_term",
                    impact=EnergyScenarioImpact(
                        cost_reduction=12.4,  # Still achieves savings
                        carbon_emissions=2120,  # Maintains carbon reduction
                        community_health=0.85,  # Protects vulnerable communities
                        worker_safety=0.92,  # High safety protection
                        energy_justice=0.87,  # Excellent energy justice
                        long_term_sustainability=0.92,
                        grid_stability=0.88
                    ),
                    value_aligned_score=0.0,
                    compliance_status="FULL_COMPLIANCE",
                    ethical_violations=[]
                ),
                
                # Scenario 3: Community-First Approach
                EthicalEnergyScenario(
                    id="community_first",
                    name="Community-First Approach",
                    description="Maximum community protection with moderate energy goals",
                    cost_increase=18.5,
                    implementation_time="long_term",
                    impact=EnergyScenarioImpact(
                        cost_reduction=8.7,
                        carbon_emissions=2450,
                        community_health=0.95,
                        worker_safety=0.94,
                        energy_justice=0.92,
                        long_term_sustainability=0.78,
                        grid_stability=0.82
                    ),
                    value_aligned_score=0.0,
                    compliance_status="FULL_COMPLIANCE",
                    ethical_violations=[]
                ),
                
                # Scenario 4: Environmental Excellence
                EthicalEnergyScenario(
                    id="environmental_excellence",
                    name="Environmental Excellence",
                    description="Maximum environmental protection with ethical considerations",
                    cost_increase=22.3,
                    implementation_time="long_term",
                    impact=EnergyScenarioImpact(
                        cost_reduction=15.8,
                        carbon_emissions=1780,  # 37.5% reduction
                        community_health=0.88,
                        worker_safety=0.90,
                        energy_justice=0.85,
                        long_term_sustainability=0.95,
                        grid_stability=0.90
                    ),
                    value_aligned_score=0.0,
                    compliance_status="FULL_COMPLIANCE",
                    ethical_violations=[]
                ),
                
                # Scenario 5: Balanced Sustainability
                EthicalEnergyScenario(
                    id="balanced_sustainability",
                    name="Balanced Sustainability",
                    description="Optimal balance across all ethical dimensions",
                    cost_increase=15.2,
                    implementation_time="medium_term",
                    impact=EnergyScenarioImpact(
                        cost_reduction=18.5,
                        carbon_emissions=2320,
                        community_health=0.82,
                        worker_safety=0.88,
                        energy_justice=0.80,
                        long_term_sustainability=0.85,
                        grid_stability=0.86
                    ),
                    value_aligned_score=0.0,
                    compliance_status="FULL_COMPLIANCE",
                    ethical_violations=[]
                )
            ]
            
            self.scenarios = scenarios
            return scenarios
        
        def check_ethical_compliance(self, scenario: EthicalEnergyScenario) -> Tuple[str, List[str]]:
            """Check if scenario complies with ethical constraints"""
            violations = []
            
            for constraint in self.constraints:
                if constraint.principle == EthicalPrinciple.WORKER_SAFETY:
                    if scenario.impact.worker_safety < constraint.threshold:
                        violations.append(f"Worker safety: {scenario.impact.worker_safety:.2f} < {constraint.threshold:.2f}")
                
                elif constraint.principle == EthicalPrinciple.COMMUNITY_WELFARE:
                    if scenario.impact.community_health < constraint.threshold - 0.1:  # Allow some flexibility
                        violations.append(f"Community welfare: {scenario.impact.community_health:.2f} < {constraint.threshold - 0.1:.2f}")
                
                elif constraint.principle == EthicalPrinciple.ENVIRONMENTAL_PROTECTION:
                    if scenario.impact.carbon_emissions > constraint.threshold:
                        violations.append(f"Carbon emissions: {scenario.impact.carbon_emissions:.0f} > {constraint.threshold:.0f} tCO₂")
                
                elif constraint.principle == EthicalPrinciple.ENERGY_JUSTICE:
                    if scenario.impact.energy_justice < constraint.threshold:
                        violations.append(f"Energy justice: {scenario.impact.energy_justice:.2f} < {constraint.threshold:.2f}")
                
                elif constraint.principle == EthicalPrinciple.ECONOMIC_SUSTAINABILITY:
                    if scenario.cost_increase > abs(constraint.threshold) * 100:
                        violations.append(f"Economic sustainability: {scenario.cost_increase:.1f}% > {abs(constraint.threshold)*100:.1f}%")
                
                elif constraint.principle == EthicalPrinciple.GRID_STABILITY:
                    if scenario.impact.grid_stability < constraint.threshold:
                        violations.append(f"Grid stability: {scenario.impact.grid_stability:.2f} < {constraint.threshold:.2f}")
            
            # Check for minimum cost reduction
            if scenario.impact.cost_reduction < 5.0:  # Minimum 5% cost reduction
                violations.append(f"Insufficient cost reduction: {scenario.impact.cost_reduction:.1f}% < 5.0%")
            
            compliance_status = "VIOLATION" if violations else "FULL_COMPLIANCE"
            return compliance_status, violations
        
        def calculate_value_aligned_score(self, scenario: EthicalEnergyScenario) -> float:
            """Calculate value-aligned score for a scenario"""
            
            # Check compliance first
            compliance_status, violations = self.check_ethical_compliance(scenario)
            
            # If hard constraints violated, score is 0
            hard_constraint_violations = [v for v in violations if any(c.principle.value in v and c.is_hard_constraint for c in self.constraints)]
            if hard_constraint_violations:
                return 0.0
            
            # Calculate weighted stakeholder score
            stakeholder_scores = []
            
            for stakeholder in self.stakeholders:
                if stakeholder.name == "Community":
                    score = (scenario.impact.community_health * 0.4 + 
                            scenario.impact.energy_justice * 0.4 + 
                            scenario.impact.grid_stability * 0.2)
                
                elif stakeholder.name == "Workers":
                    score = scenario.impact.worker_safety
                
                elif stakeholder.name == "Environment":
                    # Fixed syntax error: convert boolean to int properly
                    carbon_target_met = 1.0 if scenario.impact.carbon_emissions <= 2320 else 0.0
                    score = carbon_target_met * 0.5  # Carbon target met
                    score += scenario.impact.long_term_sustainability * 0.3
                    score += scenario.impact.energy_justice * 0.2
                
                elif stakeholder.name == "Operations":
                    score = (scenario.impact.cost_reduction / 25.3 * 0.5 +  # Normalized to max reduction
                            scenario.impact.grid_stability * 0.3 + 
                            (1 - abs(scenario.cost_increase) / 50) * 0.2)  # Cost reasonableness
                
                stakeholder_scores.append(score * stakeholder.weight)
            
            # Apply penalty for soft constraint violations
            soft_constraint_violations = [v for v in violations if not any(c.principle.value in v and c.is_hard_constraint for c in self.constraints)]
            penalty = len(soft_constraint_violations) * 0.15
            
            # Apply vulnerability bonus for protecting vulnerable groups
            vulnerability_bonus = 0.0
            if scenario.impact.community_health > 0.8 and scenario.impact.worker_safety > 0.85:
                vulnerability_bonus = 0.1
            
            # Calculate final score
            base_score = sum(stakeholder_scores)
            final_score = max(0.0, base_score - penalty + vulnerability_bonus)
            
            return min(1.0, final_score)  # Cap at 1.0
        
        def evaluate_scenarios(self) -> List[EthicalEnergyScenario]:
            """Evaluate all scenarios and calculate scores"""
            
            for scenario in self.scenarios:
                # Calculate value-aligned score
                scenario.value_aligned_score = self.calculate_value_aligned_score(scenario)
                
                # Check compliance and update status
                compliance_status, violations = self.check_ethical_compliance(scenario)
                scenario.compliance_status = compliance_status
                scenario.ethical_violations = violations
            
            return self.scenarios
    
    # Create the value-aligned energy management system
    ethical_energy_system = ValueAlignedEnergyManagement()
    scenarios = ethical_energy_system.create_ethical_scenarios()
    scenarios = ethical_energy_system.evaluate_scenarios()
    
    print(f"Created {len(scenarios)} ethical energy scenarios for evaluation")
    print(f"Stakeholder groups: {[s.name for s in ethical_energy_system.stakeholders]}")
    print(f"Ethical constraints: {len(ethical_energy_system.constraints)}")
    print(f"Carbon budget: {ethical_energy_system.carbon_budget} tCO₂/year")
    print(f"Reduction target: {ethical_energy_system.reduction_target*100:.1f}%")
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    def analyze_ethical_energy_scenarios(scenarios: List[EthicalEnergyScenario]) -> Dict:
        """Analyze and display ethical scenario evaluation results"""
    
        print("\n" + "=" * 80)
        print("VALUE-ALIGNED ENERGY MANAGEMENT RESULTS")
        print("=" * 80)
    
        # Sort scenarios by value-aligned score
        sorted_scenarios = sorted(scenarios, key=lambda x: x.value_aligned_score, reverse=True)
    
        print(f"\n--- SCENARIO RANKING BY VALUE-ALIGNED SCORE ---")
        print(f"{'Rank':<6} {'Scenario':<20} {'Score':<8} {'Compliance':<15} {'Cost Reduction':<18}")
        print("-" * 75)
    
        for i, scenario in enumerate(sorted_scenarios):
            print(f"{i+1:<6} {scenario.name[:18]:<20} {scenario.value_aligned_score:<8.3f} "
                  f"{scenario.compliance_status:<15} {scenario.impact.cost_reduction:>8.1f}%")
    
        # Detailed analysis of top scenarios
        print(f"\n--- DETAILED ANALYSIS OF TOP SCENARIOS ---")
    
        for scenario in sorted_scenarios[:3]:
            print(f"\n{scenario.name}:")
            print(f"  Description: {scenario.description}")
            print(f"  Value-Aligned Score: {scenario.value_aligned_score:.3f}")
            print(f"  Ethical Compliance: {scenario.compliance_status}")
            print(f"  Cost Impact: {scenario.cost_increase:+.1f}%")
            print(f"  Implementation: {scenario.implementation_time}")
    
            print(f"  Stakeholder Impacts:")
            print(f"    Cost Reduction: {scenario.impact.cost_reduction:+.1f}%")
            print(f"    Carbon Emissions: {scenario.impact.carbon_emissions:.0f} tCO₂/year")
            print(f"    Community Health: {scenario.impact.community_health:+.2f}")
            print(f"    Worker Safety: {scenario.impact.worker_safety:+.2f}")
            print(f"    Energy Justice: {scenario.impact.energy_justice:+.2f}")
            print(f"    Long-term Sustainability: {scenario.impact.long_term_sustainability:+.2f}")
            print(f"    Grid Stability: {scenario.impact.grid_stability:+.2f}")
    
            # Check compliance details
            if scenario.ethical_violations:
                print(f"  Ethical Violations:")
                for violation in scenario.ethical_violations:
                    print(f"    - {violation}")
    
        # Compare with expected results (from source material)
        print(f"\n--- COMPARISON WITH SOURCE MATERIAL EXPECTATIONS ---")
    
        # Find the value-aligned scenario (should match source material)
        value_aligned_scenario = next((s for s in scenarios if s.id == "value_aligned"), None)
        if value_aligned_scenario:
            print(f"Value-Aligned Scenario (from source material):")
            print(f"  Expected cost increase: 12.4%")
            print(f"  Actual cost increase: {value_aligned_scenario.cost_increase:+.1f}%")
            print(f"  Expected community protection: 100%")
            print(f"  Actual community health: {value_aligned_scenario.impact.community_health:.2f}")
            print(f"  Expected worker safety: High (9.2/10)")
            print(f"  Actual worker safety: {value_aligned_scenario.impact.worker_safety:.2f}")
            print(f"  Expected carbon reduction: 18.3%")
            actual_carbon_reduction = (2850 - value_aligned_scenario.impact.carbon_emissions) / 2850
            print(f"  Actual carbon reduction: {actual_carbon_reduction*100:.1f}%")
    
        # Select best scenario
        best_scenario = sorted_scenarios[0]
    
        print(f"\n--- ETHICAL DECISION RECOMMENDATION ---")
        print(f"Recommended Scenario: {best_scenario.name}")
        print(f"Rationale:")
        print(f"  - Highest value-aligned score: {best_scenario.value_aligned_score:.3f}")
        print(f"  - Ethical compliance: {best_scenario.compliance_status}")
        print(f"  - Cost reduction: {best_scenario.impact.cost_reduction:.1f}%")
        print(f"  - Carbon emissions: {best_scenario.impact.carbon_emissions:.0f} tCO₂/year")
        print(f"  - Community protection: {best_scenario.impact.community_health:.2f}")
        print(f"  - Worker safety: {best_scenario.impact.worker_safety:.2f}")
        print(f"  - Energy justice: {best_scenario.impact.energy_justice:.2f}")
    
        return {
            'best_scenario': best_scenario,
            'all_scenarios': sorted_scenarios,
            'ethical_compliance': {s.id: s.compliance_status for s in scenarios},
            'value_aligned_scores': {s.id: s.value_aligned_score for s in scenarios}
        }
    
    # Analyze ethical energy scenarios
    analysis_results = analyze_ethical_energy_scenarios(scenarios)
except Exception as e:
    print(f'Error in cell: {e}')
```

```python
try:
    def visualize_ethical_energy_analysis(scenarios: List[EthicalEnergyScenario], analysis_results: Dict):
        """Create comprehensive visualizations of ethical energy analysis"""
        
        fig, axes = plt.subplots(2, 2, figsize=(16, 12))
        fig.suptitle('Value-Aligned & Ethical Energy Management Analysis', fontsize=16, fontweight='bold')
        
        # 1. Value-aligned scores comparison
        ax1 = axes[0, 0]
        
        scenario_names = [s.name[:15] for s in scenarios]
        scores = [s.value_aligned_score for s in scenarios]
        colors = ['lightgreen' if s.compliance_status == 'FULL_COMPLIANCE' else 'lightcoral' for s in scenarios]
        
        bars = ax1.bar(scenario_names, scores, color=colors, alpha=0.8, edgecolor='black')
        ax1.set_xlabel('Scenarios')
        ax1.set_ylabel('Value-Aligned Score')
        ax1.set_title('Value-Aligned Scores by Scenario')
        ax1.set_ylim(0, 1)
        ax1.grid(True, alpha=0.3, axis='y')
        
        # Add value labels on bars
        for bar, score in zip(bars, scores):
            height = bar.get_height()
            ax1.text(bar.get_x() + bar.get_width()/2, height + 0.01,
                    f'{score:.3f}', ha='center', va='bottom', fontweight='bold')
        
        # Add compliance indicator
        for i, scenario in enumerate(scenarios):
            if scenario.compliance_status == 'VIOLATION':
                ax1.text(i, 0.05, '✗', ha='center', va='center', fontsize=20, color='red', fontweight='bold')
            else:
                ax1.text(i, 0.05, '✓', ha='center', va='center', fontsize=20, color='green', fontweight='bold')
        
        # Rotate x-axis labels
        plt.setp(ax1.get_xticklabels(), rotation=45, ha='right')
        
        # 2. Cost reduction vs carbon emissions
        ax2 = axes[0, 1]
        
        cost_reductions = [s.impact.cost_reduction for s in scenarios]
        carbon_emissions = [s.impact.carbon_emissions for s in scenarios]
        community_scores = [s.impact.community_health for s in scenarios]
        
        scatter = ax2.scatter(cost_reductions, carbon_emissions, 
                           c=community_scores, s=100, alpha=0.7, cmap='RdYlGn', edgecolors='black')
        
        # Add scenario labels
        for i, scenario in enumerate(scenarios):
            ax2.annotate(scenario.name[:10], (cost_reductions[i], carbon_emissions[i]), 
                        xytext=(5, 5), textcoords='offset points', fontsize=8)
        
        ax2.set_xlabel('Cost Reduction (%)')
        ax2.set_ylabel('Carbon Emissions (tCO₂/year)')
        ax2.set_title('Cost Reduction vs Carbon Emissions')
        ax2.grid(True, alpha=0.3)
        ax2.invert_yaxis()  # Lower carbon emissions are better
        
        # Add colorbar for community health
        cbar = plt.colorbar(scatter, ax=ax2)
        cbar.set_label('Community Health Impact')
        
        # Add reference lines
        ax2.axhline(y=2320, color='blue', linestyle='--', alpha=0.5, label='Carbon Target')
        ax2.axvline(x=12.4, color='green', linestyle='--', alpha=0.5, label='Expected Cost Impact')
        ax2.legend()
        
        # 3. Stakeholder impact radar chart
        ax3 = axes[1, 0]
        
        # Select top 3 scenarios for radar chart
        top_scenarios = sorted(scenarios, key=lambda x: x.value_aligned_score, reverse=True)[:3]
        
        categories = ['Community\nHealth', 'Worker\nSafety', 'Energy\nJustice', 
                      'Long-term\nSustainability', 'Grid\nStability']
        
        # Create radar chart
        angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
        angles += angles[:1]  # Complete the circle
        
        ax3 = plt.subplot(2, 2, 3, projection='polar')
        
        colors = ['blue', 'green', 'orange']
        for i, scenario in enumerate(top_scenarios):
            values = [
                scenario.impact.community_health,
                scenario.impact.worker_safety,
                scenario.impact.energy_justice,
                scenario.impact.long_term_sustainability,
                scenario.impact.grid_stability
            ]
            values += values[:1]  # Complete the circle
            
            ax3.plot(angles, values, 'o-', linewidth=2, label=scenario.name[:12], color=colors[i])
            ax3.fill(angles, values, alpha=0.25, color=colors[i])
        
        ax3.set_xticks(angles[:-1])
        ax3.set_xticklabels(categories)
        ax3.set_ylim(0, 1)
        ax3.set_title('Stakeholder Impact Comparison\n(Top 3 Scenarios)', pad=20)
        ax3.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
        ax3.grid(True)
        
        # 4. Cost vs Compliance with ethical constraints
        ax4 = axes[1, 1]
        
        costs = [s.cost_increase for s in scenarios]
        value_scores = [s.value_aligned_score for s in scenarios]
        compliance_status = [s.compliance_status for s in scenarios]
        
        # Create scatter plot with different markers for compliance
        compliant_scenarios = [(c, v) for c, v, comp in zip(costs, value_scores, compliance_status) if comp == 'FULL_COMPLIANCE']
        non_compliant_scenarios = [(c, v) for c, v, comp in zip(costs, value_scores, compliance_status) if comp == 'VIOLATION']
        
        if compliant_scenarios:
            compliant_costs, compliant_scores = zip(*compliant_scenarios)
            ax4.scatter(compliant_costs, compliant_scores, s=100, c='green', alpha=0.7, 
                       edgecolors='black', linewidth=2, label='Full Compliance', marker='o')
        
        if non_compliant_scenarios:
            non_compliant_costs, non_compliant_scores = zip(*non_compliant_scenarios)
            ax4.scatter(non_compliant_costs, non_compliant_scores, s=100, c='red', alpha=0.7, 
                       edgecolors='black', linewidth=2, label='Violation', marker='X')
        
        # Add scenario labels
        for i, scenario in enumerate(scenarios):
            ax4.annotate(scenario.name[:10], (costs[i], value_scores[i]), 
                        xytext=(5, 5), textcoords='offset points', fontsize=8)
        
        ax4.set_xlabel('Cost Increase (%)')
        ax4.set_ylabel('Value-Aligned Score')
        ax4.set_title('Cost vs Value-Aligned Score\n(by Compliance Status)')
        ax4.grid(True, alpha=0.3)
        ax4.legend()
        
        # Add reference lines
        ax4.axvline(x=12.4, color='blue', linestyle='--', alpha=0.5, label='Expected Cost Impact')
        ax4.axhline(y=0.78, color='green', linestyle='--', alpha=0.5, label='Expected Score')
        
        plt.tight_layout()
        plt.show()
    
    # Visualize ethical energy analysis
    visualize_ethical_energy_analysis(scenarios, analysis_results)
except Exception as e:
    print(f'Error in cell: {e}')
```
