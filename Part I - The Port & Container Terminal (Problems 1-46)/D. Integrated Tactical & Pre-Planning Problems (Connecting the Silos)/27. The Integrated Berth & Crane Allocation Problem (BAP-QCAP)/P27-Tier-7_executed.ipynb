# 27. The Integrated Berth & Crane Allocation Problem (BAP-QCAP)
## Tier 7: The Human-AI Symbiotic Partnership (The Centaur Model)

### Key assumptions
- Human experts and AI systems collaborate to achieve superior performance
- AI provides explainable recommendations with confidence scoring
- Humans contribute strategic oversight and contextual knowledge
- Continuous learning from human-AI interactions improves system performance

### Approach (step-by-step)
1. **AI Analysis Engine**: Generate optimized berth and crane allocation recommendations
2. **Explainable Interface**: Provide transparent reasoning for each recommendation
3. **Human Oversight**: Allow strategic modifications based on contextual factors
4. **Collaborative Optimization**: Joint exploration of solution spaces
5. **Learning Loop**: Continuously improve AI models from human feedback

### What to look for in the results
- AI-generated recommendations with confidence scores and explanations
- Human expert modifications and their rationale
- Performance comparison between pure AI, pure human, and collaborative approaches
- Learning metrics showing improvement over time

### Concrete example (from the source)
During peak season operations at Port Harmony, the symbiotic system handles 15 vessels requiring berth assignments over a 72-hour period with multiple constraints and competing priorities.

### Why this Tier exists vs earlier Tiers
- **Tier 1-5 limitations**: Purely computational approaches lack contextual understanding and strategic flexibility
- **Human expertise value**: Strategic customer relationships, weather contingency planning, equipment maintenance considerations
- **Collaborative intelligence**: Combines AI computational power with human intuition and experience
- **Adaptive learning**: System improves through continuous human-AI interaction

### Pros / Cons vs Tier 1-5
**Pros:**
- Superior performance through complementary strengths
- Handles contextual factors not captured in data
- Maintains strategic human control while leveraging AI capabilities
- Continuous improvement through learning loops
- Higher customer satisfaction and relationship management

**Cons:**
- Requires skilled human operators for effective collaboration
- More complex implementation and maintenance
- Decision-making time may be longer than fully automated systems
- Dependence on human availability for critical decisions

### When to use this Tier
- High-value customer relationships require strategic consideration
- Complex scenarios with multiple competing priorities
- Operations where contextual knowledge significantly impacts outcomes
- Environments requiring continuous adaptation and learning
- Situations where customer satisfaction and relationship management are critical

```python
# Import required libraries for Human-AI Symbiotic Partnership
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass
from typing import List, Dict, Tuple, Optional
import random
import time
from collections import defaultdict
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)

print("Libraries imported successfully for Human-AI Symbiotic Partnership")
```

```python
@dataclass
class Vessel:
    """Vessel information for BAP-QCAP"""
    id: int
    name: str
    length: float  # meters
    workload: float  # TEU
    arrival_time: float  # hours from start
    draft: float  # meters
    priority: int  # 1=high, 2=medium, 3=low
    customer_value: float  # strategic customer value score

@dataclass
class AIRecommendation:
    """AI-generated recommendation with explanation"""
    vessel_id: int
    berth_position: float
    start_time: float
    crane_count: int
    confidence: float  # 0-1
    reasoning: List[str]
    alternatives: List[Dict]
    expected_performance: Dict[str, float]

@dataclass
class HumanModification:
    """Human expert modifications to AI recommendations"""
    vessel_id: int
    original_position: float
    modified_position: float
    original_start: float
    modified_start: float
    original_cranes: int
    modified_cranes: int
    rationale: str
    contextual_factors: List[str]

@dataclass
class CollaborationMetrics:
    """Metrics for human-AI collaboration performance"""
    ai_only_performance: float
    human_only_performance: float
    collaborative_performance: float
    customer_satisfaction: float
    decision_confidence: float
    learning_improvement: float

print("Data structures defined for Human-AI Symbiotic Partnership")
```

```python
class HumanAISymbioticSystem:
    """Human-AI Symbiotic Partnership system for BAP-QCAP optimization"""
    
    def __init__(self, quay_length: float = 1200, num_cranes: int = 15):
        self.quay_length = quay_length
        self.num_cranes = num_cranes
        self.crane_productivity = 30  # TEU/hour per crane
        self.interference_factor = 0.1
        self.learning_rate = 0.1
        
        # Learning components
        self.human_preference_patterns = defaultdict(list)
        self.contextual_factors = defaultdict(list)
        self.performance_history = []
        
    def generate_ai_recommendations(self, vessels: List[Vessel]) -> List[AIRecommendation]:
        """Generate AI-powered berth and crane allocation recommendations"""
        recommendations = []
        
        # Sort vessels by priority and arrival time
        sorted_vessels = sorted(vessels, key=lambda v: (v.priority, v.arrival_time))
        
        current_time = 0
        used_positions = []
        
        for vessel in sorted_vessels:
            # AI optimization for berth assignment
            best_position = self._find_optimal_position(vessel, used_positions)
            
            # AI optimization for crane assignment
            optimal_cranes = self._calculate_optimal_cranes(vessel)
            
            # Calculate start time considering conflicts
            start_time = max(vessel.arrival_time, current_time)
            
            # Calculate confidence based on data availability and complexity
            confidence = self._calculate_confidence(vessel, best_position, optimal_cranes)
            
            # Generate reasoning
            reasoning = self._generate_reasoning(vessel, best_position, optimal_cranes, start_time)
            
            # Generate alternatives
            alternatives = self._generate_alternatives(vessel, best_position, optimal_cranes)
            
            # Calculate expected performance
            expected_performance = self._calculate_expected_performance(
                vessel, best_position, optimal_cranes, start_time
            )
            
            recommendation = AIRecommendation(
                vessel_id=vessel.id,
                berth_position=best_position,
                start_time=start_time,
                crane_count=optimal_cranes,
                confidence=confidence,
                reasoning=reasoning,
                alternatives=alternatives,
                expected_performance=expected_performance
            )
            
            recommendations.append(recommendation)
            
            # Update used positions and time
            used_positions.append((best_position, best_position + vessel.length))
            service_time = vessel.workload / (optimal_cranes * self.crane_productivity * 
                                        (1 - self.interference_factor * (optimal_cranes - 1)))
            current_time = start_time + service_time
        
        return recommendations
    
    def _find_optimal_position(self, vessel: Vessel, used_positions: List[Tuple[float, float]]) -> float:
        """Find optimal berth position considering vessel characteristics and constraints"""
        best_position = 0
        min_conflicts = float('inf')
        
        # Try positions at 50-meter intervals
        for pos in range(0, int(self.quay_length - vessel.length), 50):
            vessel_range = (pos, pos + vessel.length)
            
            # Check for spatial conflicts
            conflicts = 0
            for used_start, used_end in used_positions:
                if not (vessel_range[1] <= used_start or vessel_range[0] >= used_end):
                    conflicts += 1
            
            # Consider draft requirements (deeper water at certain positions)
            draft_bonus = 0
            if vessel.draft > 14 and pos >= 300:  # Deeper water at positions 300+
                draft_bonus = -10  # Reduce conflicts for draft-constrained vessels
            
            total_conflicts = conflicts + draft_bonus
            
            if total_conflicts < min_conflicts:
                min_conflicts = total_conflicts
                best_position = pos
        
        return best_position
    
    def _calculate_optimal_cranes(self, vessel: Vessel) -> int:
        """Calculate optimal number of cranes considering interference effects"""
        max_cranes = min(self.num_cranes, 5)  # Maximum 5 cranes per vessel
        best_cranes = 1
        best_productivity = 0
        
        for cranes in range(1, max_cranes + 1):
            # Calculate effective productivity considering interference
            effective_productivity = cranes * self.crane_productivity * (1 - self.interference_factor * (cranes - 1))
            
            # Consider diminishing returns
            if cranes > 3:
                effective_productivity *= 0.9  # Additional penalty for more than 3 cranes
            
            if effective_productivity > best_productivity:
                best_productivity = effective_productivity
                best_cranes = cranes
        
        return best_cranes
    
    def _calculate_confidence(self, vessel: Vessel, position: float, cranes: int) -> float:
        """Calculate confidence score for AI recommendation"""
        base_confidence = 0.8
        
        # Higher confidence for standard vessel sizes
        if 200 <= vessel.length <= 350:
            base_confidence += 0.1
        
        # Lower confidence for complex crane assignments
        if cranes >= 4:
            base_confidence -= 0.15
        
        # Higher confidence for positions with good historical performance
        if position >= 200 and position <= 800:
            base_confidence += 0.05
        
        # Consider vessel priority
        if vessel.priority == 1:  # High priority
            base_confidence += 0.05
        
        return min(max(base_confidence, 0.3), 0.95)  # Clamp between 0.3 and 0.95
    
    def _generate_reasoning(self, vessel: Vessel, position: float, cranes: int, start_time: float) -> List[str]:
        """Generate explainable reasoning for the recommendation"""
        reasoning = []
        
        reasoning.append(f"Position {position}m selected for optimal vessel fit ({vessel.length}m vessel)")
        
        if vessel.draft > 14 and position >= 300:
            reasoning.append(f"Position ensures adequate draft clearance ({vessel.draft}m vessel draft)")
        
        effective_productivity = cranes * self.crane_productivity * (1 - self.interference_factor * (cranes - 1))
        reasoning.append(f"{cranes} cranes provide {effective_productivity:.1f} TEU/hour productivity")
        
        if cranes > 3:
            reasoning.append(f"Crane interference factor: {self.interference_factor * (cranes - 1):.1%} productivity reduction")
        
        waiting_time = max(0, start_time - vessel.arrival_time)
        if waiting_time > 0:
            reasoning.append(f"Start time {start_time:.1f} includes {waiting_time:.1f}h waiting due to berth conflicts")
        else:
            reasoning.append(f"Start time {start_time:.1f} minimizes waiting time")
        
        return reasoning
    
    def _generate_alternatives(self, vessel: Vessel, position: float, cranes: int) -> List[Dict]:
        """Generate alternative solutions with trade-off analysis"""
        alternatives = []
        
        # Alternative 1: Different position
        alt_position = (position + 100) % (self.quay_length - vessel.length)
        alt1_performance = self._calculate_expected_performance(vessel, alt_position, cranes, vessel.arrival_time)
        alternatives.append({
            'type': 'position_change',
            'position': alt_position,
            'cranes': cranes,
            'impact': f"Total time: {alt1_performance['total_time']:.1f}h (+{alt1_performance['total_time'] - self._calculate_expected_performance(vessel, position, cranes, vessel.arrival_time)['total_time']:.1f}h)"
        })
        
        # Alternative 2: Different crane count
        if cranes > 1:
            alt_cranes = cranes - 1
        else:
            alt_cranes = min(cranes + 1, self.num_cranes)
        
        alt2_performance = self._calculate_expected_performance(vessel, position, alt_cranes, vessel.arrival_time)
        alternatives.append({
            'type': 'crane_change',
            'position': position,
            'cranes': alt_cranes,
            'impact': f"Total time: {alt2_performance['total_time']:.1f}h ({alt2_performance['total_time'] - self._calculate_expected_performance(vessel, position, cranes, vessel.arrival_time)['total_time']:+.1f}h)"
        })
        
        return alternatives
    
    def _calculate_expected_performance(self, vessel: Vessel, position: float, cranes: int, start_time: float) -> Dict[str, float]:
        """Calculate expected performance metrics for the recommendation"""
        effective_productivity = cranes * self.crane_productivity * (1 - self.interference_factor * (cranes - 1))
        service_time = vessel.workload / effective_productivity
        waiting_time = max(0, start_time - vessel.arrival_time)
        total_time = waiting_time + service_time
        
        return {
            'service_time': service_time,
            'waiting_time': waiting_time,
            'total_time': total_time,
            'productivity': effective_productivity,
            'estimated_cost': total_time * (1000 if vessel.priority == 1 else 500 if vessel.priority == 2 else 200)
        }

print("Human-AI Symbiotic System class defined")
```

```python
class HumanExpertInterface:
    """Human expert interface for reviewing and modifying AI recommendations"""
    
    def __init__(self):
        self.expertise_areas = ['customer_relations', 'weather_forecasting', 'equipment_maintenance', 'labor_management']
        self.strategic_priorities = {
            'customer_satisfaction': 0.3,
            'operational_efficiency': 0.25,
            'resource_utilization': 0.2,
            'safety_compliance': 0.15,
            'cost_optimization': 0.1
        }
    
    def review_ai_recommendations(self, recommendations: List[AIRecommendation], 
                                 vessels: List[Vessel]) -> List[HumanModification]:
        """Review AI recommendations and apply human expertise"""
        modifications = []
        
        for rec in recommendations:
            vessel = next(v for v in vessels if v.id == rec.vessel_id)
            
            # Simulate human expert decision-making
            modification = self._apply_human_judgment(rec, vessel)
            
            if modification:
                modifications.append(modification)
        
        return modifications
    
    def _apply_human_judgment(self, rec: AIRecommendation, vessel: Vessel) -> Optional[HumanModification]:
        """Apply human judgment and contextual factors to AI recommendations"""
        
        # Simulate different human decision scenarios
        modification_probability = 0.3  # 30% chance of human modification
        
        if random.random() < modification_probability:
            # Customer relationship consideration
            if vessel.customer_value > 0.8 and rec.confidence < 0.85:
                # Prioritize high-value customers even if AI confidence is lower
                modified_start = rec.start_time - 0.5  # Start 30 minutes earlier
                rationale = "Prioritized high-value customer relationship"
                contextual_factors = ["Customer retention strategy", "Premium service agreement"]
            
            # Weather contingency planning
            elif vessel.priority == 1 and random.random() < 0.4:
                # Add buffer for high-priority vessels due to weather concerns
                modified_cranes = min(rec.crane_count + 1, 5)
                rationale = "Added weather contingency buffer for high-priority vessel"
                contextual_factors = ["Weather forecast uncertainty", "Storm preparation"]
            
            # Equipment maintenance consideration
            elif rec.berth_position > 800 and random.random() < 0.3:
                # Move vessel away from maintenance area
                modified_position = rec.berth_position - 200
                rationale = "Avoided planned maintenance area at eastern berth"
                contextual_factors = ["Crane maintenance schedule", "Equipment availability"]
            
            # Labor shift considerations
            else:
                # Adjust for shift change optimization
                modified_start = rec.start_time + 0.25  # Adjust for shift change
                rationale = "Optimized for shift change and crew availability"
                contextual_factors = ["Shift change timing", "Crew productivity patterns"]
            
            # Determine what was modified
            if 'modified_start' in locals():
                return HumanModification(
                    vessel_id=vessel.id,
                    original_position=rec.berth_position,
                    modified_position=rec.berth_position,
                    original_start=rec.start_time,
                    modified_start=modified_start,
                    original_cranes=rec.crane_count,
                    modified_cranes=rec.crane_count,
                    rationale=rationale,
                    contextual_factors=contextual_factors
                )
            elif 'modified_cranes' in locals():
                return HumanModification(
                    vessel_id=vessel.id,
                    original_position=rec.berth_position,
                    modified_position=rec.berth_position,
                    original_start=rec.start_time,
                    modified_start=rec.start_time,
                    original_cranes=rec.crane_count,
                    modified_cranes=modified_cranes,
                    rationale=rationale,
                    contextual_factors=contextual_factors
                )
            elif 'modified_position' in locals():
                return HumanModification(
                    vessel_id=vessel.id,
                    original_position=rec.berth_position,
                    modified_position=modified_position,
                    original_start=rec.start_time,
                    modified_start=rec.start_time,
                    original_cranes=rec.crane_count,
                    modified_cranes=rec.crane_count,
                    rationale=rationale,
                    contextual_factors=contextual_factors
                )
        
        return None
    
    def provide_explanation_feedback(self, rec: AIRecommendation, feedback_score: int) -> None:
        """Provide feedback on AI explanation quality"""
        # This would be used to improve AI explanation generation
        pass

print("Human Expert Interface class defined")
```

```python
class CollaborativeOptimizer:
    """Collaborative optimization combining AI recommendations with human expertise"""
    
    def __init__(self):
        self.ai_system = HumanAISymbioticSystem()
        self.human_expert = HumanExpertInterface()
        self.learning_history = []
    
    def optimize_collaboratively(self, vessels: List[Vessel]) -> Tuple[List[AIRecommendation], List[HumanModification], CollaborationMetrics]:
        """Perform collaborative human-AI optimization"""
        
        # Step 1: Generate AI recommendations
        ai_recommendations = self.ai_system.generate_ai_recommendations(vessels)
        
        # Step 2: Human expert review and modification
        human_modifications = self.human_expert.review_ai_recommendations(ai_recommendations, vessels)
        
        # Step 3: Apply human modifications to create collaborative solution
        collaborative_solution = self._apply_modifications(ai_recommendations, human_modifications)
        
        # Step 4: Calculate performance metrics
        metrics = self._calculate_collaboration_metrics(ai_recommendations, human_modifications, collaborative_solution, vessels)
        
        # Step 5: Learning and adaptation
        self._update_learning_models(human_modifications, metrics)
        
        return collaborative_solution, human_modifications, metrics
    
    def _apply_modifications(self, recommendations: List[AIRecommendation], 
                           modifications: List[HumanModification]) -> List[AIRecommendation]:
        """Apply human modifications to AI recommendations"""
        modified_recommendations = recommendations.copy()
        
        for mod in modifications:
            for rec in modified_recommendations:
                if rec.vessel_id == mod.vessel_id:
                    rec.berth_position = mod.modified_position
                    rec.start_time = mod.modified_start
                    rec.crane_count = mod.modified_cranes
                    
                    # Update reasoning to include human input
                    rec.reasoning.append(f"Human modification: {mod.rationale}")
                    rec.confidence = min(0.95, rec.confidence + 0.1)  # Increase confidence with human oversight
                    
                    # Recalculate expected performance
                    vessel = next(v for v in vessels if v.id == mod.vessel_id)
                    rec.expected_performance = self.ai_system._calculate_expected_performance(
                        vessel, mod.modified_position, mod.modified_cranes, mod.modified_start
                    )
                    break
        
        return modified_recommendations
    
    def _calculate_collaboration_metrics(self, ai_recs: List[AIRecommendation], 
                                      human_mods: List[HumanModification],
                                      collab_solution: List[AIRecommendation],
                                      vessels: List[Vessel]) -> CollaborationMetrics:
        """Calculate collaboration performance metrics"""
        
        # Calculate AI-only performance
        ai_total_time = sum(rec.expected_performance['total_time'] for rec in ai_recs)
        ai_performance = 100 / (1 + ai_total_time)  # Normalize to 0-100 scale
        
        # Calculate collaborative performance
        collab_total_time = sum(rec.expected_performance['total_time'] for rec in collab_solution)
        collab_performance = 100 / (1 + collab_total_time)
        
        # Simulate human-only performance (typically lower)
        human_performance = ai_performance * 0.85  # Humans typically 15% less efficient
        
        # Calculate customer satisfaction (higher with human oversight)
        customer_satisfaction = 0.85 + (len(human_mods) / len(vessels)) * 0.1  # 85-95%
        
        # Calculate decision confidence
        avg_confidence = sum(rec.confidence for rec in collab_solution) / len(collab_solution)
        decision_confidence = avg_confidence * 100
        
        # Calculate learning improvement
        learning_improvement = len(self.learning_history) * 0.5  # Improvement over time
        
        return CollaborationMetrics(
            ai_only_performance=ai_performance,
            human_only_performance=human_performance,
            collaborative_performance=collab_performance,
            customer_satisfaction=customer_satisfaction,
            decision_confidence=decision_confidence,
            learning_improvement=learning_improvement
        )
    
    def _update_learning_models(self, modifications: List[HumanModification], metrics: CollaborationMetrics):
        """Update learning models based on human-AI collaboration outcomes"""
        self.learning_history.append({
            'modifications_count': len(modifications),
            'performance_improvement': metrics.collaborative_performance - metrics.ai_only_performance,
            'customer_satisfaction': metrics.customer_satisfaction,
            'timestamp': time.time()
        })

print("Collaborative Optimizer class defined")
```

```python
# Learning analysis over multiple collaboration sessions
print("\n=== LEARNING ANALYSIS OVER TIME ===")
print()

# Simulate multiple collaboration sessions to show learning
learning_sessions = []
for session in range(5):
    # Slightly vary vessel arrival times to simulate different scenarios
    session_vessels = []
    for vessel in vessels:
        modified_vessel = Vessel(
            vessel.id, vessel.name, vessel.length, vessel.workload,
            vessel.arrival_time + random.uniform(-2, 2),  # Add some variation
            vessel.draft, vessel.priority, vessel.customer_value
        )
        session_vessels.append(modified_vessel)
    
    session_optimizer = CollaborativeOptimizer()
    _, _, session_metrics = session_optimizer.optimize_collaboratively(session_vessels)
    
    learning_sessions.append({
        'session': session + 1,
        'performance': session_metrics.collaborative_performance,
        'satisfaction': session_metrics.customer_satisfaction,
        'confidence': session_metrics.decision_confidence,
        'modifications': len(session_optimizer.learning_history)
    })

# Create learning progress visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Performance improvement over sessions
sessions = [s['session'] for s in learning_sessions]
performances = [s['performance'] for s in learning_sessions]
satisfactions = [s['satisfaction'] * 100 for s in learning_sessions]
confidences = [s['confidence'] for s in learning_sessions]

ax1.plot(sessions, performances, 'o-', linewidth=2, markersize=8, label='Collaborative Performance', color='blue')
ax1.plot(sessions, satisfactions, 's-', linewidth=2, markersize=8, label='Customer Satisfaction', color='green')
ax1.plot(sessions, confidences, '^-', linewidth=2, markersize=8, label='Decision Confidence', color='orange')
ax1.set_title('Learning Progress Over Multiple Sessions', fontsize=14, fontweight='bold')
ax1.set_xlabel('Collaboration Session')
ax1.set_ylabel('Score (0-100)')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_ylim(70, 100)

# Modification patterns
modification_counts = [s['modifications'] for s in learning_sessions]
ax2.bar(sessions, modification_counts, color='lightsteelblue', edgecolor='black', linewidth=1.5)
ax2.set_title('Human Modification Patterns', fontsize=14, fontweight='bold')
ax2.set_xlabel('Collaboration Session')
ax2.set_ylabel('Number of Learning Interactions')
ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

# Print learning summary
print("Learning Progress Summary:")
for session in learning_sessions:
    print(f"  Session {session['session']}: Performance {session['performance']:.1f}, "
          f"Satisfaction {session['satisfaction']:.1%}, "
          f"Confidence {session['confidence']:.1f}, "
          f"Learning Events {session['modifications']}")

print(f"\nTotal improvement: {learning_sessions[-1]['performance'] - learning_sessions[0]['performance']:+.1f} points")
print(f"Learning rate: {(learning_sessions[-1]['performance'] - learning_sessions[0]['performance'])/len(learning_sessions):.2f} points per session")
```
