2. **Communication Protocol**: Implement message passing and negotiation mechanisms
3. **Market Mechanism**: Create auction-based resource allocation system
4. **Coordination Strategies**: Develop distributed decision-making algorithms
5. **Emergent Behavior**: Observe system-wide optimization from local interactions

```python
# Multi-Agent System Core Components
@dataclass
class Message:
    """Message for agent communication"""
    sender_id: str
    receiver_id: str
    message_type: str
    content: dict
    timestamp: float
    priority: int = 1

@dataclass
class Auction:
    """Auction for resource allocation"""
    auction_id: str
    resource_type: str
    resource: dict
    bidders: List[str]
    current_bid: Optional[dict]
    deadline: float
    winner: Optional[str] = None
    status: str = 'active'

class Agent:
    """Base class for autonomous agents"""
    def __init__(self, agent_id: str, agent_type: str):
        self.agent_id = agent_id
        self.agent_type = agent_type
        
        # BDI (Belief-Desire-Intention) architecture
        self.beliefs = {}
        self.desires = []
        self.intentions = []
        
        # Communication
        self.inbox = deque(maxlen=100)
        self.outbox = []
        self.message_history = deque(maxlen=1000)
        
        # State
        self.is_active = True
        self.current_task = None
        self.performance_history = deque(maxlen=100)
        
    def send_message(self, receiver_id: str, message_type: str, content: dict, priority: int = 1):
        """Send message to another agent"""
        message = Message(
            sender_id=self.agent_id,
            receiver_id=receiver_id,
            message_type=message_type,
            content=content,
            timestamp=time.time(),
            priority=priority
        )
        self.outbox.append(message)
        self.message_history.append(message)
    
    def receive_message(self, message: Message):
        """Receive message from another agent"""
        self.inbox.append(message)
        self.message_history.append(message)
    
    def process_messages(self):
        """Process incoming messages"""
        while self.inbox:
            message = self.inbox.popleft()
            self.handle_message(message)
    
    def handle_message(self, message: Message):
        """Handle incoming message (to be overridden by subclasses)"""
        pass
    
    def update_beliefs(self, new_beliefs: dict):
        """Update agent's beliefs about the environment"""
        self.beliefs.update(new_beliefs)
    
    def add_desire(self, desire: dict):
        """Add a new desire/goal"""
        self.desires.append(desire)
    
    def form_intentions(self):
        """Form intentions based on beliefs and desires"""
        self.intentions = []
        for desire in self.desires:
            if self.can_achieve(desire):
                self.intentions.append(desire)
    
    def can_achieve(self, desire: dict) -> bool:
        """Check if agent can achieve a desire"""
        return True
    
    def act(self):
        """Execute current intentions"""
        if self.intentions:
            intention = self.intentions[0]
            self.execute_intention(intention)
    
    def execute_intention(self, intention: dict):
        """Execute a specific intention (to be overridden)"""
        pass
```

```python
# Specialized Agent Classes
class VehicleAgent(Agent):
    """Autonomous vehicle agent"""
    def __init__(self, agent_id: str, capacity: int, cost_per_km: float, initial_position: str):
        super().__init__(agent_id, 'vehicle')
        
        # Vehicle properties
        self.capacity = capacity
        self.cost_per_km = cost_per_km
        self.current_position = initial_position
        self.current_load = 0
        self.fuel_level = 100.0
        self.total_distance = 0.0
        self.assigned_containers = []
        
        # Economic preferences
        self.min_profit_margin = 0.1
        self.risk_tolerance = 0.5
        
        # Initialize beliefs and desires
        self.update_beliefs({
            'position': initial_position,
            'capacity': capacity,
            'current_load': 0,
            'fuel_level': 100.0,
            'market_prices': {},
        })
        
        self.add_desire({'type': 'maximize_profit', 'priority': 1})
        self.add_desire({'type': 'maintain_utilization', 'target': 0.8, 'priority': 2})
        self.add_desire({'type': 'minimize_empty_runs', 'priority': 3})
    
    def handle_message(self, message: Message):
        """Handle incoming messages"""
        if message.message_type == 'auction_announcement':
            self.evaluate_auction(message.content)
        elif message.message_type == 'assignment':
            self.accept_assignment(message.content)
        elif message.message_type == 'market_update':
            self.update_beliefs(message.content)
    
    def evaluate_auction(self, auction_info: dict):
        """Evaluate if auction is worth participating"""
        resource = auction_info['resource']
        
        if resource['type'] == 'container':
            # Check if vehicle can handle the container
            if (self.current_load + resource['weight'] <= self.capacity and
                self.fuel_level > 20 and self.is_active):
                
                # Calculate bid price
                distance = self.calculate_distance(resource['origin'], resource['destination'])
                base_cost = distance * self.cost_per_km
                bid_price = base_cost * (1 + self.min_profit_margin)
                
                # Adjust based on current utilization
                utilization = self.current_load / self.capacity
                if utilization < 0.5:
                    bid_price *= 0.9
                elif utilization > 0.8:
                    bid_price *= 1.2
                
                # Submit bid
                bid = {
                    'agent_id': self.agent_id,
                    'auction_id': auction_info['auction_id'],
                    'price': bid_price,
                    'estimated_time': distance / 25,
                    'confidence': 1.0 - utilization * self.risk_tolerance
                }
                
                self.send_message(
                    auction_info['auctioneer_id'],
                    'bid_submission',
                    bid,
                    priority=2
                )
    
    def calculate_distance(self, origin: str, destination: str) -> float:
        """Calculate distance between terminals"""
        terminal_positions = {
            'A': (0, 0), 'B': (3, 0), 'C': (1.5, 2.5), 'D': (6, 0), 'E': (4.5, 2.5)
        }
        
        if origin in terminal_positions and destination in terminal_positions:
            pos1 = terminal_positions[origin]
            pos2 = terminal_positions[destination]
            return np.sqrt((pos1[0] - pos2[0])**2 + (pos1[1] - pos2[1])**2)
        else:
            return 5.0
    
    def accept_assignment(self, assignment: dict):
        """Accept and execute container assignment"""
        container = assignment['container']
        
        # Update vehicle state
        self.assigned_containers.append(container)
        self.current_load += container['weight']
        self.current_position = container['destination']
        
        # Update beliefs
        distance = self.calculate_distance(container['origin'], container['destination'])
        self.total_distance += distance
        self.fuel_level -= distance * 0.1
        
        self.update_beliefs({
            'current_load': self.current_load,
            'current_position': self.current_position,
            'total_distance': self.total_distance,
            'fuel_level': self.fuel_level,
            'assigned_containers': len(self.assigned_containers)
        })
        
        # Record performance
        profit = assignment['price'] - (distance * self.cost_per_km)
        self.performance_history.append({
            'timestamp': time.time(),
            'action': 'assignment',
            'profit': profit,
            'utilization': self.current_load / self.capacity
        })

class TerminalAgent(Agent):
    """Terminal management agent"""
    def __init__(self, agent_id: str, terminal_id: str, capacity: int):
        super().__init__(agent_id, 'terminal')
        
        # Terminal properties
        self.terminal_id = terminal_id
        self.capacity = capacity
        self.current_containers = []
        self.processing_rate = 10
        self.queue_length = 0
        self.congestion_level = 0.0
        
        # Market management
        self.active_auctions = {}
        self.local_market_prices = {}
        
        # Initialize beliefs and desires
        self.update_beliefs({
            'terminal_id': terminal_id,
            'capacity': capacity,
            'current_containers': 0,
            'processing_rate': self.processing_rate,
            'congestion_level': 0.0,
        })
        
        self.add_desire({'type': 'minimize_congestion', 'priority': 1})
        self.add_desire({'type': 'maximize_throughput', 'priority': 2})
        self.add_desire({'type': 'balance_workload', 'priority': 3})
    
    def handle_message(self, message: Message):
        """Handle incoming messages"""
        if message.message_type == 'container_arrival':
            self.process_container_arrival(message.content)
        elif message.message_type == 'bid_submission':
            self.process_bid(message.content)
        elif message.message_type == 'market_request':
            self.provide_market_update(message.content)
    
    def process_container_arrival(self, container_info: dict):
        """Process new container arrival"""
        container = container_info['container']
        self.current_containers.append(container)
        self.queue_length = len(self.current_containers)
        
        # Update congestion level
        self.congestion_level = min(1.0, self.queue_length / self.capacity)
        
        # Create auction for container transfer
        if self.congestion_level > 0.7:
            self.create_container_auction(container)
        
        # Update beliefs
        self.update_beliefs({
            'current_containers': len(self.current_containers),
            'queue_length': self.queue_length,
            'congestion_level': self.congestion_level,
        })
    
    def create_container_auction(self, container: dict):
        """Create auction for container transfer"""
        auction_id = f"auction_{self.terminal_id}_{container['id']}_{int(time.time())}"
        
        auction = Auction(
            auction_id=auction_id,
            resource_type='container',
            resource=container,
            bidders=[],
            current_bid=None,
            deadline=time.time() + 30,
        )
        
        self.active_auctions[auction_id] = auction
        
        # Announce auction to market
        auction_info = {
            'auction_id': auction_id,
            'auctioneer_id': self.agent_id,
            'resource': container,
            'deadline': auction.deadline,
        }
        
        self.send_message(
            'coordinator',
            'auction_announcement',
            auction_info,
            priority=3
        )
    
    def process_bid(self, bid_info: dict):
        """Process bid for auction"""
        auction_id = bid_info['auction_id']
        
        if auction_id in self.active_auctions:
            auction = self.active_auctions[auction_id]
            
            # Check if bid is better than current bid
            if (auction.current_bid is None or 
                bid_info['price'] < auction.current_bid['price']):
                
                auction.current_bid = bid_info
                auction.bidders.append(bid_info['agent_id'])
                
                # Update market price
                self.local_market_prices[auction.resource['id']] = bid_info['price']
    
    def close_auction(self, auction_id: str):
        """Close auction and award resource"""
        if auction_id in self.active_auctions:
            auction = self.active_auctions[auction_id]
            
            if auction.current_bid:
                # Award to winner
                winner_id = auction.current_bid['agent_id']
                auction.winner = winner_id
                
                # Send assignment
                assignment = {
                    'container': auction.resource,
                    'price': auction.current_bid['price'],
                    'winner': winner_id,
                }
                
                self.send_message(
                    winner_id,
                    'assignment',
                    assignment,
                    priority=3
                )
                
                # Remove container from queue
                self.current_containers.remove(auction.resource)
                self.queue_length = len(self.current_containers)
                
                # Record transaction
                self.performance_history.append({
                    'timestamp': time.time(),
                    'action': 'auction_completed',
                    'price': auction.current_bid['price'],
                    'winner': winner_id,
                    'congestion_level': self.congestion_level,
                })
            
            # Close auction
            auction.status = 'closed'
            del self.active_auctions[auction_id]
    
    def act(self):
        """Execute terminal operations"""
        # Process containers
        if self.current_containers and self.congestion_level < 0.8:
            processed = min(self.processing_rate // 60, len(self.current_containers))
            for _ in range(processed):
                self.current_containers.pop(0)
            
            self.queue_length = len(self.current_containers)
            self.congestion_level = min(1.0, self.queue_length / self.capacity)
        
        # Check and close expired auctions
        expired_auctions = [
            aid for aid, auction in self.active_auctions.items()
            if time.time() > auction.deadline
        ]
        
        for auction_id in expired_auctions:
            self.close_auction(auction_id)
        
        # Update beliefs
        self.update_beliefs({
            'current_containers': len(self.current_containers),
            'queue_length': self.queue_length,
            'congestion_level': self.congestion_level,
        })

class CoordinatorAgent(Agent):
    """Market coordinator agent"""
    def __init__(self, agent_id: str):
        super().__init__(agent_id, 'coordinator')
        
        # Market management
        self.registered_agents = {}
        self.market_state = {
            'total_containers': 0,
            'active_vehicles': 0,
            'average_prices': {},
            'market_efficiency': 0.0,
        }
        
        # Initialize beliefs and desires
        self.update_beliefs({
            'registered_agents': {},
            'market_state': self.market_state,
        })
        
        self.add_desire({'type': 'market_efficiency', 'priority': 1})
        self.add_desire({'type': 'price_stability', 'priority': 2})
        self.add_desire({'type': 'fair_allocation', 'priority': 3})
    
    def handle_message(self, message: Message):
        """Handle incoming messages"""
        if message.message_type == 'registration':
            self.register_agent(message.content)
        elif message.message_type == 'auction_announcement':
            self.distribute_auction(message.content)
        elif message.message_type == 'bid_submission':
            self.forward_bid(message.content)
        elif message.message_type == 'market_update':
            self.update_market_state(message.content)
    
    def register_agent(self, agent_info: dict):
        """Register agent in the market"""
        agent_id = agent_info['agent_id']
        agent_type = agent_info['agent_type']
        
        self.registered_agents[agent_id] = {
            'type': agent_type,
            'capabilities': agent_info.get('capabilities', {}),
            'status': 'active',
            'registration_time': time.time(),
        }
        
        # Update market state
        if agent_type == 'vehicle':
            self.market_state['active_vehicles'] += 1
        
        self.update_beliefs({
            'registered_agents': self.registered_agents,
            'market_state': self.market_state,
        })
    
    def distribute_auction(self, auction_info: dict):
        """Distribute auction announcement to relevant agents"""
        # Send to all vehicle agents
        for agent_id, agent_info in self.registered_agents.items():
            if agent_info['type'] == 'vehicle' and agent_info['status'] == 'active':
                self.send_message(
                    agent_id,
                    'auction_announcement',
                    auction_info,
                    priority=2
                )
    
    def forward_bid(self, bid_info: dict):
        """Forward bid to auctioneer"""
        auctioneer_id = bid_info.get('auctioneer_id')
        if auctioneer_id and auctioneer_id in self.registered_agents:
            self.send_message(
                auctioneer_id,
                'bid_submission',
                bid_info,
                priority=2
            )
    
    def update_market_state(self, market_info: dict):
        """Update market state information"""
        terminal_id = market_info.get('terminal_id')
        
        if terminal_id:
            self.market_state[f'terminal_{terminal_id}'] = market_info
        
        # Update aggregate metrics
        total_containers = 0
        for key, value in self.market_state.items():
            if key.startswith('terminal_'):
                total_containers += value.get('queue_length', 0)
        
        self.market_state['total_containers'] = total_containers
        
        # Calculate market efficiency
        total_capacity = 0
        for agent_id, agent_info in self.registered_agents.items():
            if agent_info['type'] == 'vehicle':
                total_capacity += agent_info['capabilities'].get('capacity', 0)
        
        if total_capacity > 0:
            self.market_state['market_efficiency'] = total_containers / total_capacity
        
        self.update_beliefs({
            'market_state': self.market_state,
        })
    
    def act(self):
        """Coordinate market operations"""
        # Request market updates from terminals
        for agent_id, agent_info in self.registered_agents.items():
            if agent_info['type'] == 'terminal':
                self.send_message(
                    agent_id,
                    'market_request',
                    {'requester_id': self.agent_id},
                    priority=1
                )
        
        # Analyze and optimize market
        self.optimize_market()
    
    def optimize_market(self):
        """Optimize market operations"""
        # Check for market inefficiencies
        if self.market_state['market_efficiency'] < 0.5:
            # Market is inefficient - take corrective action
            self.stimulate_market()
    
    def stimulate_market(self):
        """Stimulate market activity"""
        # Could implement various stimulation strategies
        pass
```

```python
# Multi-Agent System Implementation
class InterTerminalMultiAgentSystem:
    """Complete multi-agent system for inter-terminal operations"""
    def __init__(self, terminals, vehicles, distances):
        self.terminals = terminals
        self.vehicles = vehicles
        self.distances = distances
        
        # Create agents
        self.agents = {}
        self.message_queue = deque(maxlen=1000)
        
        # System state
        self.current_time = 0
        self.simulation_step = 0
        
        # Performance tracking
        self.system_metrics = deque(maxlen=1000)
        self.auction_history = deque(maxlen=500)
        
        # Initialize the system
        self._initialize_agents()
        self._initialize_containers()
    
    def _initialize_agents(self):
        """Initialize all agents in the system"""
        # Create coordinator agent
        coordinator = CoordinatorAgent('coordinator_001')
        self.agents[coordinator.agent_id] = coordinator
        
        # Create terminal agents
        for i, terminal in enumerate(self.terminals):
            terminal_agent = TerminalAgent(
                agent_id=f'terminal_{terminal}',
                terminal_id=terminal,
                capacity=50
            )
            self.agents[terminal_agent.agent_id] = terminal_agent
        
        # Create vehicle agents
        for i, vehicle in enumerate(self.vehicles):
            vehicle_agent = VehicleAgent(
                agent_id=f'vehicle_{vehicle.id}',
                capacity=vehicle.capacity,
                cost_per_km=vehicle.cost_per_km,
                initial_position=random.choice(self.terminals)
            )
            self.agents[vehicle_agent.agent_id] = vehicle_agent
        
        # Register agents with coordinator
        for agent_id, agent in self.agents.items():
            if agent.agent_type != 'coordinator':
                registration_info = {
                    'agent_id': agent.agent_id,
                    'agent_type': agent.agent_type,
                    'capabilities': self._get_agent_capabilities(agent)
                }
                
                agent.send_message(
                    'coordinator_001',
                    'registration',
                    registration_info,
                    priority=3
                )
    
    def _get_agent_capabilities(self, agent):
        """Get agent capabilities"""
        if agent.agent_type == 'vehicle':
            return {
                'capacity': agent.capacity,
                'cost_per_km': agent.cost_per_km,
                'current_position': agent.current_position,
                'current_load': agent.current_load,
                'fuel_level': agent.fuel_level,
            }
        elif agent.agent_type == 'terminal':
            return {
                'terminal_id': agent.terminal_id,
                'capacity': agent.capacity,
                'processing_rate': agent.processing_rate,
                'queue_length': agent.queue_length,
                'congestion_level': agent.congestion_level,
            }
        else:
            return {}
    
    def _initialize_containers(self):
        """Initialize containers in the system"""
        # Create initial containers
        container_id = 1
        
        for _ in range(20):
            container = {
                'id': container_id,
                'origin': random.choice(self.terminals),
                'destination': random.choice([t for t in self.terminals if t != random.choice(self.terminals)]),
                'weight': random.choice([1, 2, 3]),
                'priority': random.choice(['low', 'medium', 'high']),
                'ready_time': self.current_time,
                'status': 'waiting',
                'type': 'container'
            }
            
            # Assign to random terminal
            terminal_id = random.choice(self.terminals)
            terminal_agent = self.agents[f'terminal_{terminal_id}']
            
            terminal_agent.send_message(
                terminal_agent.agent_id,
                'container_arrival',
                {'container': container},
                priority=2
            )
            
            container_id += 1
    
    def simulate_step(self, time_step=60):
        """Simulate one time step of the multi-agent system"""
        self.current_time += time_step
        self.simulation_step += 1
        
        # Process messages
        self._process_messages()
        
        # Agent actions
        self._execute_agent_actions()
        
        # Generate new containers
        if self.simulation_step % 10 == 0:
            self._generate_new_containers()
        
        # Collect system metrics
        self._collect_system_metrics()
        
        return self._get_system_state()
    
    def _process_messages(self):
        """Process all messages in the queue"""
        while self.message_queue:
            message = self.message_queue.popleft()
            
            # Deliver message to recipient
            if message.receiver_id in self.agents:
                self.agents[message.receiver_id].receive_message(message)
    
    def _execute_agent_actions(self):
        """Execute actions for all agents"""
        for agent in self.agents.values():
            if agent.is_active:
                agent.process_messages()
                agent.form_intentions()
                agent.act()
                
                # Send outgoing messages
                while agent.outbox:
                    message = agent.outbox.popleft()
                    self.message_queue.append(message)
    
    def _generate_new_containers(self):
        """Generate new containers in the system"""
        num_containers = random.randint(1, 3)
        
        for _ in range(num_containers):
            container_id = len([c for agent in self.agents.values() 
                                if agent.agent_type == 'terminal']) + 
                              sum(len(ta.current_containers) for ta in self.agents.values() 
                                if ta.agent_type == 'terminal') + 1
            
            container = {
                'id': container_id,
                'origin': random.choice(self.terminals),
                'destination': random.choice([t for t in self.terminals if t != random.choice(self.terminals)]),
                'weight': random.choice([1, 2, 3]),
                'priority': random.choice(['low', 'medium', 'high']),
                'ready_time': self.current_time,
                'status': 'waiting',
                'type': 'container'
            }
            
            # Assign to random terminal
            terminal_id = random.choice(self.terminals)
            terminal_agent = self.agents[f'terminal_{terminal_id}']
            
            terminal_agent.send_message(
                terminal_agent.agent_id,
                'container_arrival',
                {'container': container},
                priority=2
            )
    
    def _collect_system_metrics(self):
        """Collect system-wide performance metrics"""
        total_containers = sum(len(ta.current_containers) for ta in self.agents.values() 
                                if ta.agent_type == 'terminal')
        
        total_capacity = sum(va.capacity for va in self.agents.values() 
                               if va.agent_type == 'vehicle')
        
        total_load = sum(va.current_load for va in self.agents.values() 
                           if va.agent_type == 'vehicle')
        
        active_auctions = sum(len(ta.active_auctions) for ta in self.agents.values() 
                               if ta.agent_type == 'terminal')
        
        # Calculate efficiency metrics
        system_efficiency = total_load / max(1, total_capacity)
        
        # Calculate auction success rate
        completed_auctions = 0
        total_auctions = 0
        for ta in self.agents.values():
            if ta.agent_type == 'terminal':
                for performance in ta.performance_history:
                    if performance['action'] == 'auction_completed':
                        completed_auctions += 1
                        total_auctions += 1
        
        auction_success_rate = completed_auctions / max(1, total_auctions)
        
        metrics = {
            'timestamp': self.current_time,
            'simulation_step': self.simulation_step,
            'total_containers': total_containers,
            'total_capacity': total_capacity,
            'total_load': total_load,
            'system_efficiency': system_efficiency,
            'active_auctions': active_auctions,
            'auction_success_rate': auction_success_rate,
            'active_agents': len([a for a in self.agents.values() if a.is_active]),
        }
        
        self.system_metrics.append(metrics)
    
    def _get_system_state(self):
        """Get current system state for analysis"""
        return {
            'current_time': self.current_time,
            'simulation_step': self.simulation_step,
            'agents': {
                agent_id: {
                    'type': agent.agent_type,
                    'is_active': agent.is_active,
                    'beliefs': agent.beliefs,
                    'performance_history': list(agent.performance_history)
                }
                for agent_id, agent in self.agents.items()
            },
            'metrics': list(self.system_metrics)[-1] if self.system_metrics else {},
        }
    
    def run_simulation(self, num_steps=100):
        """Run the multi-agent simulation"""
        print("=== MULTI-AGENT SYSTEM SIMULATION ===")
        print(f"Agents: {len(self.agents)} (Vehicles: {len([a for a in self.agents.values() if a.agent_type == 'vehicle'])}, Terminals: {len([a for a in self.agents.values() if a.agent_type == 'terminal'])})")
        print(f"Simulation steps: {num_steps}")
        print(f"Time step: 60 seconds per step")
        
        start_time = time.time()
        
        for step in range(num_steps):
            state = self.simulate_step()
            
            # Progress reporting
            if step % 10 == 0 or step == num_steps - 1:
                metrics = state['metrics']
                print(f"Step {step:3d}: "
                      f"Containers: {metrics.get('total_containers', 0):3d}, "
                      f"Load: {metrics.get('total_load', 0):3d}/{metrics.get('total_capacity', 0):3d} ({metrics.get('system_efficiency', 0):.1%}), "
                      f"Auctions: {metrics.get('active_auctions', 0):2d}, "
                      f"Success Rate: {metrics.get('auction_success_rate', 0):.1%}")
        
        simulation_time = time.time() - start_time
        
        print(f"\nSimulation completed in {simulation_time:.2f} seconds")
        return self._analyze_results()
    
    def _analyze_results(self):
        """Analyze simulation results"""
        if not self.system_metrics:
            return {}
        
        # Calculate final metrics
        final_metrics = self.system_metrics[-1]
        
        # Calculate averages
        avg_efficiency = np.mean([m['system_efficiency'] for m in self.system_metrics])
        avg_success_rate = np.mean([m['auction_success_rate'] for m in self.system_metrics])
        avg_containers = np.mean([m['total_containers'] for m in self.system_metrics])
        
        # Calculate agent performance
        vehicle_performance = []
        terminal_performance = []
        
        for agent in self.agents.values():
            if agent.agent_type == 'vehicle' and agent.performance_history:
                profits = [p['profit'] for p in agent.performance_history if 'profit' in p]
                if profits:
                    vehicle_performance.append({
                        'agent_id': agent.agent_id,
                        'avg_profit': np.mean(profits),
                        'total_assignments': len(agent.assigned_containers),
                        'utilization': agent.current_load / agent.capacity
                    })
            
            elif agent.agent_type == 'terminal' and agent.performance_history:
                auctions_completed = [p for p in agent.performance_history if p['action'] == 'auction_completed']
                if auctions_completed:
                    terminal_performance.append({
                        'agent_id': agent.agent_id,
                        'avg_price': np.mean([a['price'] for a in auctions_completed]),
                        'total_auctions': len(auctions_completed),
                        'avg_congestion': np.mean([a['congestion_level'] for a in auctions_completed])
                    })
        
        return {
            'final_metrics': final_metrics,
            'averages': {
                'efficiency': avg_efficiency,
                'success_rate': avg_success_rate,
                'containers': avg_containers
            },
            'vehicle_performance': vehicle_performance,
            'terminal_performance': terminal_performance,
            'simulation_steps': len(self.system_metrics),
            'metrics': list(self.system_metrics)
        }
```

```python
# Create comprehensive visualizations
def visualize_multi_agent_results(results):
    """Create comprehensive visualizations of multi-agent system results"""
    
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('Multi-Agent System Performance Analysis', fontsize=16, fontweight='bold')
    
    # 1. System Efficiency Over Time
    ax1 = axes[0, 0]
    
    steps = [m['simulation_step'] for m in results['metrics']]
    efficiency = [m['system_efficiency'] * 100 for m in results['metrics']]
    success_rate = [m['auction_success_rate'] * 100 for m in results['metrics']]
    
    ax1.plot(steps, efficiency, 'b-', linewidth=2, label='System Efficiency')
    ax1.plot(steps, success_rate, 'r-', linewidth=2, label='Auction Success Rate')
    
    ax1.set_xlabel('Simulation Step')
    ax1.set_ylabel('Performance (%)')
    ax1.set_title('System Performance Over Time')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    ax1.set_ylim([0, 100])
    
    # 2. Container Processing Dynamics
    ax2 = axes[0, 1]
    
    containers = [m['total_containers'] for m in results['metrics']]
    capacity = [m['total_capacity'] for m in results['metrics']]
    load = [m['total_load'] for m in results['metrics']]
    
    ax2.fill_between(steps, 0, capacity, alpha=0.3, color='lightblue', label='Total Capacity')
    ax2.fill_between(steps, 0, load, alpha=0.5, color='lightgreen', label='Current Load')
    ax2.plot(steps, containers, 'o-', color='red', markersize=4, label='Containers')
    
    ax2.set_xlabel('Simulation Step')
    ax2.set_ylabel('Number of Units')
    ax2.set_title('Container Processing Dynamics')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # 3. Vehicle Agent Performance
    ax3 = axes[1, 0]
    
    vehicle_data = results['vehicle_performance']
    
    if vehicle_data:
        agent_ids = [vp['agent_id'] for vp in vehicle_data]
        profits = [vp['avg_profit'] for vp in vehicle_data]
        assignments = [vp['total_assignments'] for vp in vehicle_data]
        utilizations = [vp['utilization'] * 100 for vp in vehicle_data]
        
        x = np.arange(len(agent_ids))
        width = 0.25
        
        ax3.bar(x - width, profits, width, label='Avg Profit', color='lightblue', alpha=0.8)
        ax3.bar(x, assignments, width, label='Total Assignments', color='lightgreen', alpha=0.8)
        ax3.bar(x + width, utilizations, width, label='Utilization (%)', color='lightcoral', alpha=0.8)
        
        ax3.set_xlabel('Vehicle Agent')
        ax3.set_ylabel('Value')
        ax3.set_title('Vehicle Agent Performance Comparison')
        ax3.set_xticks(x)
        ax3.set_xticklabels(agent_ids, rotation=45)
        ax3.legend()
        ax3.grid(True, alpha=0.3)
    
    # 4. Market Efficiency Analysis
    ax4 = axes[1, 1]
    
    # Calculate market efficiency over time
    efficiency_history = [m['system_efficiency'] for m in results['metrics']]
    
    # Calculate moving average
    window_size = 5
    
    moving_avg = []
    for i in range(len(efficiency_history)):
        start_idx = max(0, i - window_size + 1)
        moving_avg.append(np.mean(efficiency_history[start_idx:i+1]))
    
    ax4.fill_between(steps, 0, efficiency_history, alpha=0.3, color='lightblue', label='Instant Efficiency')
    ax4.plot(steps, moving_avg, 'r-', linewidth=3, label=f'Moving Avg ({window_size} steps)')
    
    # Add equilibrium line
    if moving_avg:
        equilibrium = np.mean(moving_avg[-10:]) if len(moving_avg) >= 10 else moving_avg[-1]
        ax4.axhline(y=equilibrium, color='green', linestyle='--', linewidth=2, label=f'Equilibrium: {equilibrium:.2f}')
    
    ax4.set_xlabel('Simulation Step')
    ax4.set_ylabel('System Efficiency')
    ax4.set_title('Market Efficiency and Convergence')
    ax4.legend()
    ax4.grid(True, alpha=0.3)
    ax4.set_ylim([0, 1])
    
    plt.tight_layout()
    plt.show()
    
    return fig

# Create visualizations
fig = visualize_multi_agent_results(results)
```
