# 34. The Hazard & IMDG Segregation Problem

## Tier 5: The Integrated Digital Twin (System-of-Systems Simulation)

### Goal
Transform IMDG segregation from isolated container placement into comprehensive maritime safety ecosystem simulation, integrating vessel dynamics, weather forecasting, port infrastructure, and real-time cargo monitoring through digital twin technology.

### Key assumptions
- Physical vessel has IoT sensors monitoring container integrity and environmental conditions
- Digital model maintains real-time 3D representation with physics simulation
- Analytics engine processes sensor data to predict segregation risk in real-time
- System can generate what-if scenarios for proactive decision support

### Approach (step-by-step)
1. Create digital twin architecture with physical, digital, and analytics components
2. Implement vessel dynamics simulator with motion and cargo shifting forces
3. Build environmental monitor for weather and sea conditions
4. Develop cargo integrity tracking with IoT sensor simulation
5. Create predictive risk engine for segregation failure probability
6. Generate what-if scenarios with risk-cost analysis

### What to look for in the results
- Real-time risk monitoring with probability assessments
- Multi-sensor data integration showing system-wide situational awareness
- What-if scenario analysis with measurable trade-offs
- Predictive analytics forecasting potential segregation failures
- Automated response recommendations with cost-benefit analysis

### Concrete example (from the source)
During a North Atlantic crossing, the MV Atlantic Explorer Digital Twin detects through accelerometer data that severe weather is causing excessive container movement in Hold 3, where Class 1.4 explosives are stowed 4 meters from Class 8.1 corrosives. Temperature sensors show the corrosive container's internal temperature rising to 35°C (95°F), while structural stress sensors indicate the segregation barrier integrity is compromised. The Analytics Engine predicts a 23% probability of segregation failure within 6 hours and generates three response scenarios, recommending route deviation as optimal with $2.3M savings vs $47K increased fuel costs.

```python
# Define comprehensive data structures for Digital Twin
@dataclass
class Container:
    """Enhanced container with digital twin monitoring"""
    id: str
    imdg_class: str
    weight: float = 20.0
    position: Optional['Position'] = None
    # IoT sensor data
    temperature: float = 20.0  # Celsius
    pressure: float = 101.3     # kPa
    humidity: float = 50.0     # Percentage
    integrity_seal: bool = True
    vibration_level: float = 0.0  # m/s²
    last_inspection: datetime = field(default_factory=datetime.now)

@dataclass
class Position:
    """Enhanced position with structural monitoring"""
    id: str
    hold: int
    x: float
    y: float
    z: float
    # Structural integrity sensors
    structural_stress: float = 0.0  # MPa
    corrosion_level: float = 0.0   # Percentage
    fire_resistance: float = 100.0  # Percentage
    liquid_resistance: float = 100.0  # Percentage

@dataclass
class SegregationRequirement:
    """Enhanced segregation with dynamic factors"""
    class1: str
    class2: str
    requirement: str
    min_distance: float
    different_hold: bool = False
    # Dynamic factors
    weather_sensitivity: float = 1.0  # How weather affects requirement
    vibration_tolerance: float = 1.0  # Vibration tolerance factor
    temperature_factor: float = 1.0   # Temperature sensitivity

@dataclass
class VesselState:
    """Real-time vessel dynamics"""
    # Position and motion
    latitude: float = 40.7128  # Starting at NYC
    longitude: float = -74.0060
    heading: float = 45.0  # degrees
    speed: float = 15.0  # knots
    # Motion dynamics
    roll: float = 0.0      # degrees
    pitch: float = 0.0     # degrees
    yaw: float = 0.0       # degrees
    heave: float = 0.0     # meters
    # Environmental conditions
    sea_state: int = 3     # Beaufort scale 0-12
    wind_speed: float = 10.0  # knots
    wave_height: float = 1.5   # meters
    visibility: float = 10.0   # nautical miles
    air_temperature: float = 15.0  # Celsius
    water_temperature: float = 12.0  # Celsius

@dataclass
class RiskAssessment:
    """Risk assessment results"""
    failure_probability: float  # 0-1
    risk_level: str  # "Low", "Medium", "High", "Critical"
    time_to_failure: Optional[float]  # hours
    contributing_factors: List[str]
    recommended_actions: List[str]
    confidence: float  # 0-1
```

```python
# Digital Twin Core Components
class VesselDynamicsSimulator:
    """Simulates vessel motion and cargo shifting forces"""
    
    def __init__(self):
        self.vessel_state = VesselState()
        self.time_step = 0
    
    def update_dynamics(self, weather_conditions: Dict[str, float]) -> VesselState:
        """Update vessel dynamics based on weather conditions"""
        
        # Update environmental conditions
        self.vessel_state.sea_state = int(weather_conditions.get('sea_state', 3))
        self.vessel_state.wind_speed = weather_conditions.get('wind_speed', 10.0)
        self.vessel_state.wave_height = weather_conditions.get('wave_height', 1.5)
        self.vessel_state.air_temperature = weather_conditions.get('air_temp', 15.0)
        
        # Calculate vessel motion based on conditions
        wave_factor = self.vessel_state.wave_height / 3.0  # Normalize to 3m waves
        wind_factor = self.vessel_state.wind_speed / 25.0  # Normalize to 25 knots
        
        # Roll motion (side to side)
        self.vessel_state.roll = np.random.normal(0, 2.0 * wave_factor + 1.0 * wind_factor)
        self.vessel_state.roll = np.clip(self.vessel_state.roll, -15, 15)  # Limit to ±15 degrees
        
        # Pitch motion (front to back)
        self.vessel_state.pitch = np.random.normal(0, 1.5 * wave_factor + 0.5 * wind_factor)
        self.vessel_state.pitch = np.clip(self.vessel_state.pitch, -8, 8)  # Limit to ±8 degrees
        
        # Heave (vertical motion)
        self.vessel_state.heave = np.random.normal(0, 0.5 * wave_factor)
        self.vessel_state.heave = np.clip(self.vessel_state.heave, -2, 2)  # Limit to ±2 meters
        
        # Update position (simplified navigation)
        self.vessel_state.latitude += 0.01 * np.cos(np.radians(self.vessel_state.heading))
        self.vessel_state.longitude += 0.01 * np.sin(np.radians(self.vessel_state.heading))
        
        # Add some heading variation
        self.vessel_state.heading += np.random.normal(0, 0.5 * wind_factor)
        
        self.time_step += 1
        return self.vessel_state
    
    def calculate_cargo_forces(self, container: Container, position: Position) -> Dict[str, float]:
        """Calculate forces acting on container due to vessel motion"""
        
        # Simplified force calculation
        roll_force = abs(self.vessel_state.roll) * container.weight * 0.1  # Lateral force
        pitch_force = abs(self.vessel_state.pitch) * container.weight * 0.05  # Longitudinal force
        heave_force = abs(self.vessel_state.heave) * container.weight * 0.2   # Vertical force
        
        # Add position-dependent factors
        height_factor = 1.0 + (position.z / 20.0)  # Higher positions experience more force
        
        total_force = (roll_force + pitch_force + heave_force) * height_factor
        
        return {
            'roll_force': roll_force,
            'pitch_force': pitch_force,
            'heave_force': heave_force,
            'total_force': total_force,
            'vibration_level': total_force * 0.1  # Convert to vibration
        }

class EnvironmentalMonitor:
    """Monitors and predicts environmental conditions"""
    
    def __init__(self):
        self.current_conditions = {
            'sea_state': 3,
            'wind_speed': 10.0,
            'wave_height': 1.5,
            'air_temp': 15.0,
            'visibility': 10.0,
            'barometric_pressure': 1013.0
        }
        self.forecast_horizon = 72  # 72 hours
    
    def update_conditions(self, vessel_state: VesselState):
        """Update environmental conditions based on vessel state"""
        # Simulate weather changes
        if vessel_state.latitude > 45:  # North Atlantic
            self.current_conditions['sea_state'] = min(8, self.current_conditions['sea_state'] + np.random.choice([-1, 0, 1]))
            self.current_conditions['wind_speed'] = max(5, self.current_conditions['wind_speed'] + np.random.normal(0, 2))
            self.current_conditions['wave_height'] = max(0.5, self.current_conditions['wave_height'] + np.random.normal(0, 0.3))
        
        # Temperature variations
        self.current_conditions['air_temp'] += np.random.normal(0, 0.5)
        
        # Ensure realistic bounds
        self.current_conditions['sea_state'] = np.clip(self.current_conditions['sea_state'], 0, 12)
        self.current_conditions['wind_speed'] = np.clip(self.current_conditions['wind_speed'], 0, 60)
        self.current_conditions['wave_height'] = np.clip(self.current_conditions['wave_height'], 0, 15)
        self.current_conditions['air_temp'] = np.clip(self.current_conditions['air_temp'], -10, 35)
    
    def generate_forecast(self) -> List[Dict[str, float]]:
        """Generate weather forecast for next 72 hours"""
        forecast = []
        current = self.current_conditions.copy()
        
        for hour in range(self.forecast_horizon):
            # Simulate weather evolution
            if hour < 24:
                # Next 24 hours - current pattern continues
                trend = 0.1
            elif hour < 48:
                # 24-48 hours - gradual change
                trend = 0.3
            else:
                # 48-72 hours - more variable
                trend = 0.5
            
            # Update conditions with trend and randomness
            current['sea_state'] = np.clip(current['sea_state'] + np.random.normal(0, trend), 0, 12)
            current['wind_speed'] = np.clip(current['wind_speed'] + np.random.normal(0, trend * 2), 0, 60)
            current['wave_height'] = np.clip(current['wave_height'] + np.random.normal(0, trend * 0.5), 0, 15)
            
            forecast.append(current.copy())
        
        return forecast

class CargoIntegrityTracker:
    """Tracks container integrity through IoT sensors"""
    
    def __init__(self):
        self.sensor_network = True
        self.update_frequency = 30  # seconds
    
    def update_container_sensors(self, container: Container, position: Position, 
                                vessel_state: VesselState, env_conditions: Dict[str, float]) -> Container:
        """Update container sensor readings"""
        
        # Temperature changes based on environment and container type
        ambient_temp = env_conditions['air_temp']
        if container.imdg_class in ['8', '8.1']:  # Corrosives may generate heat
            container.temperature = ambient_temp + np.random.uniform(5, 15)
        elif container.imdg_class in ['3']:  # Flammable liquids
            container.temperature = ambient_temp + np.random.uniform(0, 5)
        else:
            container.temperature = ambient_temp + np.random.uniform(-2, 2)
        
        # Pressure changes with altitude and weather
        altitude_effect = position.z * 0.01  # Small pressure change with height
        weather_effect = (1013.0 - env_conditions.get('barometric_pressure', 1013.0)) * 0.1
        container.pressure = 101.3 + altitude_effect + weather_effect + np.random.normal(0, 0.5)
        
        # Humidity based on environmental conditions
        container.humidity = np.clip(50 + np.random.normal(0, 10) + (env_conditions['air_temp'] - 15) * 2, 20, 90)
        
        # Vibration from vessel motion
        vibration_base = abs(vessel_state.roll) * 0.5 + abs(vessel_state.pitch) * 0.3
        container.vibration_level = vibration_base + np.random.normal(0, 0.1)
        
        # Seal integrity may degrade with vibration and temperature
        if container.vibration_level > 2.0 or container.temperature > 35:
            container.integrity_seal = container.integrity_seal and random.random() > 0.1
        
        return container
```

```python
# Predictive Risk Engine
class PredictiveRiskEngine:
    """Analyzes multi-sensor data to predict segregation risk"""
    
    def __init__(self):
        self.risk_models = {
            'structural_failure': self._assess_structural_failure_risk,
            'segregation_breach': self._assess_segregation_breach_risk,
            'environmental_compromise': self._assess_environmental_risk,
            'sensor_anomaly': self._assess_sensor_anomaly_risk
        }
    
    def _assess_structural_failure_risk(self, containers: List[Container], 
                                        positions: List[Position], 
                                        vessel_state: VesselState) -> float:
        """Assess risk of structural failure"""
        
        max_stress = max(p.structural_stress for p in positions)
        motion_factor = (abs(vessel_state.roll) + abs(vessel_state.pitch)) / 20.0
        
        # Risk increases with stress and motion
        base_risk = max_stress / 2.0  # Normalize to 0-1 range
        motion_risk = motion_factor
        
        return min(1.0, base_risk + motion_risk)
    
    def _assess_segregation_breach_risk(self, containers: List[Container], 
                                        positions: List[Position],
                                        requirements: List[SegregationRequirement],
                                        vessel_state: VesselState) -> float:
        """Assess risk of segregation requirement breach"""
        
        # Calculate distances between containers
        from itertools import combinations
        
        total_risk = 0.0
        risk_count = 0
        
        for container1, container2 in combinations(containers, 2):
            if container1.position and container2.position:
                # Find applicable requirement
                applicable_req = None
                for req in requirements:
                    if ((req.class1 == container1.imdg_class and req.class2 == container2.imdg_class) or
                        (req.class1 == container2.imdg_class and req.class2 == container1.imdg_class)):
                        applicable_req = req
                        break
                
                if applicable_req:
                    # Calculate actual distance
                    pos1, pos2 = container1.position, container2.position
                    distance = np.sqrt((pos1.x - pos2.x)**2 + (pos1.y - pos2.y)**2 + (pos1.z - pos2.z)**2)
                    
                    # Calculate risk based on distance and dynamic factors
                    if applicable_req.different_hold:
                        if pos1.hold == pos2.hold:
                            # Same hold violation
                            risk = 0.8 * applicable_req.weather_sensitivity
                        else:
                            risk = 0.0
                    else:
                        # Distance requirement
                        if distance < applicable_req.min_distance:
                            distance_ratio = distance / applicable_req.min_distance
                            risk = (1.0 - distance_ratio) * applicable_req.weather_sensitivity
                        else:
                            risk = 0.0
                    
                    # Increase risk based on vessel motion
                    motion_amplification = 1.0 + (abs(vessel_state.roll) + abs(vessel_state.pitch)) / 30.0
                    risk *= motion_amplification
                    
                    # Increase risk based on container conditions
                    if container1.temperature > 30 or container2.temperature > 30:
                        risk *= applicable_req.temperature_factor
                    
                    if container1.vibration_level > 2.0 or container2.vibration_level > 2.0:
                        risk *= (2.0 - applicable_req.vibration_tolerance)
                    
                    total_risk += min(1.0, risk)
                    risk_count += 1
        
        return total_risk / max(1, risk_count)
    
    def _assess_environmental_risk(self, containers: List[Container], 
                                  env_conditions: Dict[str, float]) -> float:
        """Assess environmental risk factors"""
        
        risk_factors = []
        
        # High temperature risk
        high_temp_containers = sum(1 for c in containers if c.temperature > 35)
        temp_risk = min(1.0, high_temp_containers / len(containers) * 2)
        risk_factors.append(temp_risk)
        
        # High vibration risk
        high_vib_containers = sum(1 for c in containers if c.vibration_level > 2.5)
        vib_risk = min(1.0, high_vib_containers / len(containers) * 2)
        risk_factors.append(vib_risk)
        
        # Seal integrity risk
        compromised_seals = sum(1 for c in containers if not c.integrity_seal)
        seal_risk = min(1.0, compromised_seals / len(containers) * 3)
        risk_factors.append(seal_risk)
        
        # Severe weather risk
        weather_risk = min(1.0, env_conditions.get('sea_state', 3) / 8.0)
        risk_factors.append(weather_risk)
        
        return np.mean(risk_factors)
    
    def _assess_sensor_anomaly_risk(self, containers: List[Container]) -> float:
        """Assess risk based on sensor anomalies"""
        
        anomalies = 0
        
        for container in containers:
            # Check for unrealistic readings
            if container.temperature < -20 or container.temperature > 60:
                anomalies += 1
            if container.pressure < 80 or container.pressure > 120:
                anomalies += 1
            if container.humidity < 0 or container.humidity > 100:
                anomalies += 1
        
        return min(1.0, anomalies / (len(containers) * 3))
    
    def comprehensive_risk_assessment(self, containers: List[Container], 
                                       positions: List[Position],
                                       requirements: List[SegregationRequirement],
                                       vessel_state: VesselState,
                                       env_conditions: Dict[str, float]) -> RiskAssessment:
        """Perform comprehensive risk assessment"""
        
        # Calculate individual risk components
        risks = {}
        for risk_name, risk_model in self.risk_models.items():
            if risk_name == 'structural_failure':
                risks[risk_name] = risk_model(containers, positions, vessel_state)
            elif risk_name == 'segregation_breach':
                risks[risk_name] = risk_model(containers, positions, requirements, vessel_state)
            elif risk_name == 'environmental_compromise':
                risks[risk_name] = risk_model(containers, env_conditions)
            elif risk_name == 'sensor_anomaly':
                risks[risk_name] = risk_model(containers)
        
        # Calculate overall failure probability
        # Weight different risk types
        weights = {
            'structural_failure': 0.2,
            'segregation_breach': 0.4,
            'environmental_compromise': 0.3,
            'sensor_anomaly': 0.1
        }
        
        failure_probability = sum(risks[risk] * weights[risk] for risk in risks.keys())
        
        # Determine risk level
        if failure_probability < 0.1:
            risk_level = "Low"
            time_to_failure = None
        elif failure_probability < 0.3:
            risk_level = "Medium"
            time_to_failure = np.random.uniform(24, 72)  # 1-3 days
        elif failure_probability < 0.6:
            risk_level = "High"
            time_to_failure = np.random.uniform(6, 24)  # 6-24 hours
        else:
            risk_level = "Critical"
            time_to_failure = np.random.uniform(1, 6)   # 1-6 hours
        
        # Identify contributing factors
        contributing_factors = []
        for risk_name, risk_value in risks.items():
            if risk_value > 0.3:
                contributing_factors.append(f"{risk_name.replace('_', ' ').title()}: {risk_value:.1%}")
        
        # Generate recommended actions
        recommended_actions = []
        if risks['segregation_breach'] > 0.3:
            recommended_actions.append("Review container placement and consider re-stowage")
        if risks['structural_failure'] > 0.3:
            recommended_actions.append("Reduce vessel speed and change course to minimize motion")
        if risks['environmental_compromise'] > 0.3:
            recommended_actions.append("Inspect container seals and cooling systems")
        if risks['sensor_anomaly'] > 0.2:
            recommended_actions.append("Calibrate or replace malfunctioning sensors")
        
        # Calculate confidence based on data quality
        confidence = 1.0 - risks['sensor_anomaly'] * 0.5  # Reduce confidence if sensors are anomalous
        
        return RiskAssessment(
            failure_probability=failure_probability,
            risk_level=risk_level,
            time_to_failure=time_to_failure,
            contributing_factors=contributing_factors,
            recommended_actions=recommended_actions,
            confidence=confidence
        )
```

```python
# Digital Twin Integration System
class DigitalTwinSystem:
    """Main digital twin integration system"""
    
    def __init__(self, containers: List[Container], positions: List[Position], 
                 requirements: List[SegregationRequirement]):
        self.containers = containers
        self.positions = positions
        self.requirements = requirements
        
        # Initialize core components
        self.vessel_simulator = VesselDynamicsSimulator()
        self.env_monitor = EnvironmentalMonitor()
        self.cargo_tracker = CargoIntegrityTracker()
        self.risk_engine = PredictiveRiskEngine()
        
        # System state
        self.current_time = datetime.now()
        self.sync_interval = 30  # seconds
        self.risk_history = []
        
        # Place containers in initial configuration
        self._place_containers_initially()
    
    def _place_containers_initially(self):
        """Place containers in initial configuration"""
        # Simple placement for demonstration - in reality this would be optimized
        available_positions = self.positions.copy()
        
        for container in self.containers:
            if available_positions:
                # Prioritize placement based on IMDG class
                if container.imdg_class in ['1.1', '1.4S']:
                    # Explosives in hold 1 (isolated)
                    hold_1_positions = [p for p in available_positions if p.hold == 1]
                    if hold_1_positions:
                        position = random.choice(hold_1_positions)
                    else:
                        position = random.choice(available_positions)
                elif container.imdg_class in ['8', '8.1']:
                    # Corrosives away from explosives (hold 3 or 4)
                    far_positions = [p for p in available_positions if p.hold in [3, 4]]
                    if far_positions:
                        position = random.choice(far_positions)
                    else:
                        position = random.choice(available_positions)
                else:
                    position = random.choice(available_positions)
                
                container.position = position
                available_positions.remove(position)
    
    def synchronize(self) -> Dict[str, Any]:
        """Perform synchronization cycle between physical and digital twins"""
        
        # Update environmental conditions
        self.env_monitor.update_conditions(self.vessel_simulator.vessel_state)
        
        # Update vessel dynamics
        weather_conditions = {
            'sea_state': self.env_monitor.current_conditions['sea_state'],
            'wind_speed': self.env_monitor.current_conditions['wind_speed'],
            'wave_height': self.env_monitor.current_conditions['wave_height'],
            'air_temp': self.env_monitor.current_conditions['air_temp']
        }
        vessel_state = self.vessel_simulator.update_dynamics(weather_conditions)
        
        # Update container sensors
        for container in self.containers:
            if container.position:
                self.cargo_tracker.update_container_sensors(
                    container, container.position, vessel_state, self.env_monitor.current_conditions
                )
        
        # Perform risk assessment
        risk_assessment = self.risk_engine.comprehensive_risk_assessment(
            self.containers, self.positions, self.requirements, 
            vessel_state, self.env_monitor.current_conditions
        )
        
        # Store risk history
        self.risk_history.append({
            'timestamp': self.current_time,
            'risk_assessment': risk_assessment,
            'vessel_state': vessel_state,
            'environmental': self.env_monitor.current_conditions.copy()
        })
        
        # Keep only last 1000 records
        if len(self.risk_history) > 1000:
            self.risk_history = self.risk_history[-1000:]
        
        # Update time
        self.current_time += timedelta(seconds=self.sync_interval)
        
        return {
            'timestamp': self.current_time,
            'risk_assessment': risk_assessment,
            'vessel_state': vessel_state,
            'environmental': self.env_monitor.current_conditions,
            'sync_count': len(self.risk_history)
        }
    
    def generate_what_if_scenarios(self, risk_assessment: RiskAssessment) -> List[Dict[str, Any]]:
        """Generate what-if scenarios for decision support"""
        
        scenarios = []
        
        if risk_assessment.risk_level in ['High', 'Critical']:
            # Scenario 1: Continue current route
            scenarios.append({
                'name': 'Continue Current Route',
                'description': 'Maintain current course and speed',
                'estimated_cost': 0,
                'risk_reduction': 0.0,
                'time_impact': 0,
                'probability_of_success': risk_assessment.failure_probability
            })
            
            # Scenario 2: Reduce speed and change course
            scenarios.append({
                'name': 'Speed Reduction & Course Change',
                'description': 'Reduce speed by 50% and alter course to avoid severe weather',
                'estimated_cost': 25000,  # Additional fuel and time costs
                'risk_reduction': 0.4,  # 40% risk reduction
                'time_impact': 12,  # 12 hours delay
                'probability_of_success': max(0.1, risk_assessment.failure_probability - 0.4)
            })
            
            # Scenario 3: Emergency re-stowage at next port
            scenarios.append({
                'name': 'Emergency Re-stowage',
                'description': 'Divert to nearest port for emergency container re-stowage',
                'estimated_cost': 150000,  # Port fees, re-stowage costs, delays
                'risk_reduction': 0.9,  # 90% risk reduction
                'time_impact': 24,  # 24 hours delay
                'probability_of_success': max(0.05, risk_assessment.failure_probability - 0.9)
            })
            
            # Scenario 4: Enhanced monitoring with crew intervention
            scenarios.append({
                'name': 'Enhanced Monitoring',
                'description': 'Increase monitoring frequency and prepare crew intervention',
                'estimated_cost': 5000,  # Additional monitoring costs
                'risk_reduction': 0.2,  # 20% risk reduction through early detection
                'time_impact': 2,  # 2 hours for additional procedures
                'probability_of_success': max(0.2, risk_assessment.failure_probability - 0.2)
            })
        
        return scenarios
    
    def run_simulation(self, duration_hours: int = 24) -> Dict[str, Any]:
        """Run digital twin simulation for specified duration"""
        
        print(f"Starting Digital Twin Simulation: {duration_hours} hours")
        print("=" * 60)
        
        sync_cycles = (duration_hours * 3600) // self.sync_interval
        critical_events = []
        max_risk = 0
        
        for cycle in range(int(sync_cycles)):
            sync_result = self.synchronize()
            risk = sync_result['risk_assessment']
            
            # Track maximum risk
            if risk.failure_probability > max_risk:
                max_risk = risk.failure_probability
            
            # Check for critical events
            if risk.risk_level == 'Critical':
                critical_events.append({
                    'timestamp': sync_result['timestamp'],
                    'risk_assessment': risk,
                    'cycle': cycle
                })
                
                # Generate scenarios for first critical event
                if len(critical_events) == 1:
                    scenarios = self.generate_what_if_scenarios(risk)
                    
                    print(f"\n⚠️  CRITICAL RISK DETECTED at {sync_result['timestamp']}")
                    print(f"Failure Probability: {risk.failure_probability:.1%}")
                    print(f"Time to Failure: {risk.time_to_failure:.1f} hours" if risk.time_to_failure else "Unknown")
                    print(f"\nContributing Factors:")
                    for factor in risk.contributing_factors:
                        print(f"  - {factor}")
                    
                    print(f"\nRecommended Actions:")
                    for action in risk.recommended_actions:
                        print(f"  - {action}")
                    
                    print(f"\nWhat-if Scenarios:")
                    for scenario in scenarios:
                        expected_cost = scenario['estimated_cost'] + (scenario['probability_of_success'] * 2300000)  # Expected incident cost
                        print(f"  {scenario['name']}: ${scenario['estimated_cost']:,.0f} cost, {scenario['risk_reduction']:.0%} risk reduction, {scenario['time_impact']}h delay")
                        print(f"    Expected total cost: ${expected_cost:,.0f}")
            
            # Progress reporting
            if (cycle + 1) % 100 == 0:
                hours_elapsed = (cycle + 1) * self.sync_interval / 3600
                print(f"Simulation Progress: {hours_elapsed:.1f} hours, Risk: {risk.failure_probability:.1%} ({risk.risk_level})")
        
        return {
            'total_cycles': sync_cycles,
            'critical_events': critical_events,
            'max_risk': max_risk,
            'final_risk': self.risk_history[-1]['risk_assessment'] if self.risk_history else None,
            'scenarios': scenarios if critical_events else [],
            'risk_history': self.risk_history
        }
```

```python
# Visualize Digital Twin results
def visualize_digital_twin_results(simulation_result: Dict, digital_twin: DigitalTwinSystem):
    """Create comprehensive visualization of digital twin results"""
    
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    
    # Extract data for visualization
    risk_history = simulation_result['risk_history']
    timestamps = [r['timestamp'] for r in risk_history]
    failure_probabilities = [r['risk_assessment'].failure_probability for r in risk_history]
    vessel_states = [r['vessel_state'] for r in risk_history]
    environmental = [r['environmental'] for r in risk_history]
    
    # Plot 1: Risk probability over time
    ax1 = axes[0, 0]
    hours_elapsed = [(t - timestamps[0]).total_seconds() / 3600 for t in timestamps]
    ax1.plot(hours_elapsed, failure_probabilities, 'r-', linewidth=2)
    ax1.fill_between(hours_elapsed, 0, failure_probabilities, alpha=0.3, color='red')
    ax1.set_xlabel('Time (hours)')
    ax1.set_ylabel('Failure Probability')
    ax1.set_title('Segregation Failure Risk Over Time')
    ax1.grid(True, alpha=0.3)
    ax1.set_ylim(0, 1)
    
    # Mark critical events
    for event in simulation_result['critical_events']:
        event_time = (event['timestamp'] - timestamps[0]).total_seconds() / 3600
        ax1.scatter(event_time, event['risk_assessment'].failure_probability, 
                   color='darkred', s=100, marker='x', linewidth=3)
    
    # Plot 2: Vessel motion dynamics
    ax2 = axes[0, 1]
    rolls = [vs.roll for vs in vessel_states]
    pitches = [vs.pitch for vs in vessel_states]
    
    ax2.plot(hours_elapsed, rolls, 'b-', label='Roll', alpha=0.7, linewidth=1)
    ax2.plot(hours_elapsed, pitches, 'g-', label='Pitch', alpha=0.7, linewidth=1)
    ax2.set_xlabel('Time (hours)')
    ax2.set_ylabel('Motion (degrees)')
    ax2.set_title('Vessel Motion Dynamics')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # Plot 3: Environmental conditions
    ax3 = axes[0, 2]
    sea_states = [env['sea_state'] for env in environmental]
    wind_speeds = [env['wind_speed'] for env in environmental]
    
    ax3_twin = ax3.twinx()
    line1 = ax3.plot(hours_elapsed, sea_states, 'c-', label='Sea State', linewidth=2)
    line2 = ax3_twin.plot(hours_elapsed, wind_speeds, 'm-', label='Wind Speed', linewidth=2, alpha=0.7)
    
    ax3.set_xlabel('Time (hours)')
    ax3.set_ylabel('Sea State (Beaufort)', color='c')
    ax3_twin.set_ylabel('Wind Speed (knots)', color='m')
    ax3.set_title('Environmental Conditions')
    ax3.tick_params(axis='y', labelcolor='c')
    ax3_twin.tick_params(axis='y', labelcolor='m')
    ax3.grid(True, alpha=0.3)
    
    # Combine legends
    lines = line1 + line2
    labels = [l.get_label() for l in lines]
    ax3.legend(lines, labels, loc='upper left')
    
    # Plot 4: Container temperature distribution
    ax4 = axes[1, 0]
    final_temps = [c.temperature for c in digital_twin.containers]
    ax4.hist(final_temps, bins=15, alpha=0.7, color='orange', edgecolor='black')
    ax4.axvline(np.mean(final_temps), color='red', linestyle='--', linewidth=2, 
               label=f'Mean: {np.mean(final_temps):.1f}°C')
    ax4.set_xlabel('Temperature (°C)')
    ax4.set_ylabel('Number of Containers')
    ax4.set_title('Final Container Temperature Distribution')
    ax4.legend()
    ax4.grid(True, alpha=0.3, axis='y')
    
    # Plot 5: Risk level distribution
    ax5 = axes[1, 1]
    risk_levels = [r.risk_level for r in risk_history]
    level_counts = pd.Series(risk_levels).value_counts()
    colors = {'Low': 'green', 'Medium': 'yellow', 'High': 'orange', 'Critical': 'red'}
    bar_colors = [colors.get(level, 'gray') for level in level_counts.index]
    
    bars = ax5.bar(level_counts.index, level_counts.values, color=bar_colors, alpha=0.7)
    ax5.set_xlabel('Risk Level')
    ax5.set_ylabel('Frequency')
    ax5.set_title('Risk Level Distribution')
    ax5.grid(True, alpha=0.3, axis='y')
    
    # Add percentage labels
    total = sum(level_counts.values)
    for bar, count in zip(bars, level_counts.values):
        percentage = count / total * 100
        ax5.text(bar.get_x() + bar.get_width()/2, bar.get_height() + total*0.01, 
                f'{percentage:.1f}%', ha='center', va='bottom', fontweight='bold')
    
    # Plot 6: Scenario analysis
    ax6 = axes[1, 2]
    if simulation_result['scenarios']:
        scenarios = simulation_result['scenarios']
        scenario_names = [s['name'][:20] for s in scenarios]  # Truncate long names
        expected_costs = [s['estimated_cost'] + (s['probability_of_success'] * 2300000) for s in scenarios]
        risk_reductions = [s['risk_reduction'] for s in scenarios]
        
        # Create scatter plot of cost vs risk reduction
        scatter = ax6.scatter(expected_costs, risk_reductions, s=100, alpha=0.7, c=range(len(scenarios)), cmap='viridis')
        ax6.set_xlabel('Expected Total Cost ($)')
        ax6.set_ylabel('Risk Reduction')
        ax6.set_title('Scenario Cost-Benefit Analysis')
        ax6.grid(True, alpha=0.3)
        
        # Add scenario labels
        for i, (name, cost, reduction) in enumerate(zip(scenario_names, expected_costs, risk_reductions)):
            ax6.annotate(name, (cost, reduction), xytext=(5, 5), textcoords='offset points', fontsize=8)
        
        # Highlight best scenario
        best_idx = np.argmin(expected_costs)
        ax6.scatter(expected_costs[best_idx], risk_reductions[best_idx], s=200, 
                   facecolors='none', edgecolors='red', linewidth=3)
    else:
        ax6.text(0.5, 0.5, 'No critical events\nNo scenarios generated', 
                ha='center', va='center', transform=ax6.transAxes, fontsize=12)
        ax6.set_title('Scenario Analysis')
    
    plt.suptitle('IMDG Segregation - Digital Twin Simulation Results', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # Additional detailed analysis
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(14, 10))
    
    # Container placement by hold
    hold_distribution = {}
    for container in digital_twin.containers:
        if container.position:
            hold = container.position.hold
            hold_distribution[hold] = hold_distribution.get(hold, []) + [container]
    
    # Create hold visualization
    for hold_num in range(1, 5):
        ax = [ax1, ax2, ax3, ax4][hold_num - 1]
        hold_containers = hold_distribution.get(hold_num, [])
        hold_positions = [p for p in digital_twin.positions if p.hold == hold_num]
        
        # Draw positions
        for pos in hold_positions:
            container_at_pos = None
            for container in hold_containers:
                if container.position and container.position.id == pos.id:
                    container_at_pos = container
                    break
            
            if container_at_pos:
                # Color by IMDG class
                class_colors = {'1.1': 'darkred', '1.4S': 'red', '3': 'orange', '8': 'purple',
                               '8.1': 'darkviolet', '2.1': 'green', '2.3': 'darkgreen',
                               '5.1': 'brown', '6.1': 'olive', '4.1': 'pink', '9': 'blue'}
                color = class_colors.get(container_at_pos.imdg_class, 'gray')
                
                # Size based on vibration level
                size = 100 + container_at_pos.vibration_level * 50
                
                ax.scatter(pos.x, pos.y, s=size, c=color, marker='s', alpha=0.8, edgecolors='black')
                ax.text(pos.x, pos.y, f'{container_at_pos.id}\n{container_at_pos.imdg_class}', 
                        ha='center', va='center', fontsize=7, fontweight='bold')
            else:
                ax.scatter(pos.x, pos.y, s=100, c='lightgray', marker='s', alpha=0.3, edgecolors='black')
                ax.text(pos.x, pos.y, 'Empty', ha='center', va='center', fontsize=6)
        
        ax.set_title(f'Hold {hold_num} - Final Configuration')
        ax.set_xlabel('X Position (m)')
        ax.set_ylabel('Y Position (m)')
        ax.grid(True, alpha=0.3)
        ax.set_aspect('equal')
    
    plt.suptitle('Digital Twin - Final Container Configuration', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()

# Visualize results
visualize_digital_twin_results(simulation_result, digital_twin)
```

### Why this Tier exists vs earlier Tiers
This Tier 5 Digital Twin approach represents a paradigm shift from optimization to system integration:
- **Real-time monitoring** - Continuous surveillance vs static optimization
- **Predictive analytics** - Anticipates problems vs reactive solutions
- **System-of-systems integration** - Holistic view vs isolated container placement
- **Dynamic adaptation** - Responds to changing conditions vs fixed solutions
- **Decision support** - Provides actionable intelligence vs single solutions

### Pros / Cons vs Tiers 1-4
**Advantages over all previous Tiers:**
- **Real-time risk monitoring** with continuous assessment
- **Predictive capabilities** forecasting potential failures
- **Multi-system integration** (vessel, environment, cargo, structural)
- **Adaptive decision support** with scenario analysis
- **Comprehensive situational awareness** through sensor fusion
- **Automated alerting** for proactive risk management

**Disadvantages vs Tier 1 (MIP):**
- No optimality guarantees for initial placement
- Much higher complexity and infrastructure requirements
- Dependent on sensor reliability and data quality
- Higher computational and maintenance costs

**Disadvantages vs Tier 2 (Heuristics):**
- Requires sophisticated IoT infrastructure
- Higher implementation and operational costs
- More complex system integration challenges
- Dependent on continuous data streams

**Disadvantages vs Tier 3 (PSO):**
- Less focus on optimization excellence
- Higher infrastructure requirements
- More complex system architecture
- Dependent on real-time data availability

**Disadvantages vs Tier 4 (DRL):**
- Less adaptive learning capability
- Higher infrastructure complexity
- More focused on monitoring than learning
- Requires extensive sensor network deployment

### When to use this Tier
- **High-value cargo operations** where risk mitigation is critical
- **Regulatory compliance** requiring continuous monitoring and reporting
- **Long voyages** with changing environmental conditions
- **Complex vessels** with multiple segregation challenges
- **When real-time decision support** is more valuable than optimization
- **Operations with high incident costs** justifying monitoring investment
- **When predictive maintenance** and risk prevention are priorities
- **Integration with vessel management systems** for comprehensive oversight
